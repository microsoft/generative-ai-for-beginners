### 2. Segédkönyvtárak importálása és hitelesítő adatok példányosítása


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. A megfelelő modell megtalálása  
Az olyan modellek, mint a GPT-4o és a GPT-4o mini, képesek megérteni és előállítani természetes nyelvet, valamint elérhetők az OpenAI platformon különböző erősségű és sebességű változatokban, amelyek különböző feladatokhoz alkalmasak. 


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

#### Hogyan lehet irányítani egy nagy nyelvi modellt  
"A nagy nyelvi modell minden bemenet közül messze a legbefolyásosabb a szöveges kérés(1).

A nagy nyelvi modelleket néhány módon lehet kérésre ösztönözni:

Utasítás: Mondd meg a modellnek, mit szeretnél
Kiegészítés: Indítsd el a modellt, hogy egészítse ki a kívánt kezdetet
Bemutatás: Mutasd meg a modellnek, mit szeretnél az alábbi módokon:
Néhány példával a kérésben
Több száz vagy ezer példával egy finomhangoló tanulóadatbázisban"



#### Három alapvető irányelv van a kérés készítéséhez:

**Mutasd meg és mondd el**. Tedd világossá, mit szeretnél utasításokkal, példákkal vagy ezek kombinációjával. Ha azt szeretnéd, hogy a modell alfabetikus sorrendbe állítson egy listát vagy osztályozzon egy bekezdést érzelem szerint, mutasd meg, hogy ez a kívánt feladat.

**Biztosíts minőségi adatokat**. Ha osztályozót szeretnél építeni vagy azt szeretnéd, hogy a modell kövessen egy mintát, győződj meg róla, hogy elegendő példa van. Ellenőrizd a példáidat — a modell általában elég intelligens ahhoz, hogy átlásson az alapvető helyesírási hibákon és válaszoljon, de az is lehet, hogy feltételezi, hogy ez szándékos, ami befolyásolhatja a választ.

**Ellenőrizd a beállításaidat.** A temperature és a top_p beállítások szabályozzák, hogy a modell mennyire determinisztikusan generál választ. Ha azt kéred, hogy csak egy helyes válasz legyen, akkor ezeket alacsonyabbra állítsd. Ha változatosabb válaszokat szeretnél, akkor magasabbra érdemes állítani. Az egyik leggyakoribb hiba, hogy azt hiszik, ezek a beállítások a „ügyességet” vagy „kreativitást” szabályozzák.


Forrás: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Küldés!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
