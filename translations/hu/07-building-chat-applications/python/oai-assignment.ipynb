### 2. Segédkönyvtárak importálása és hitelesítő adatok példányosítása


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. A megfelelő modell megtalálása  
Az olyan modellek, mint a GPT-4o és a GPT-4o mini, képesek természetes nyelvet megérteni és generálni, és az OpenAI platformon érhetők el különböző teljesítmény- és sebességszinteken, amelyek különböző feladatokhoz megfelelőek.  


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Parancs Tervezés  

"A nagy nyelvi modellek varázsa abban rejlik, hogy hatalmas mennyiségű szövegen tanulva a predikciós hibát minimalizálják, így a modellek olyan fogalmakat sajátítanak el, amelyek hasznosak ezen előrejelzésekhez. Például megtanulják az olyan fogalmakat, mint"(1):

* hogyan kell helyesen írni
* hogyan működik a nyelvtan

* hogyan kell kérdésekre válaszolni
* hogyan kell beszélgetést folytatni
* hogyan kell sok nyelven írni
* hogyan kell kódolni
* stb.

#### Hogyan irányítsuk a nagy nyelvi modellt  
"A nagy nyelvi modell összes bemenete közül messze a legbefolyásosabb a szöveges parancs(1).

A nagy nyelvi modelleket néhány módon lehet parancsolni, hogy kimenetet állítsanak elő:

Utasítás: Mondd meg a modellnek, mit szeretnél
Kiegészítés: Indukáld a modellt, hogy fejezze be azt, amit elkezdtél
Bemutatás: Mutasd meg a modellnek, mit szeretnél, a következőkkel:
Néhány példa a parancsban
Több száz vagy ezer példa finomhangolási tanítóadat-készletben"



#### Három alapvető irányelv a parancsok létrehozásához:

**Mutasd meg és mondd el**. Tegye világossá, hogy mit szeretnél, akár utasításokkal, példákkal vagy ezek kombinációjával. Ha azt szeretnéd, hogy a modell egy lista elemeit ábécé sorrendbe állítsa, vagy osztályozzon egy bekezdést érzelem szerint, mutasd meg neki, hogy ez az, amit vársz.

**Adj minőségi adatokat**. Ha egy osztályozót próbálsz építeni, vagy azt akarod, hogy a modell kövessen egy mintát, győződj meg arról, hogy elég példád van. Mindig ellenőrizd a példáidat — a modell általában elég okos ahhoz, hogy átlásson az alapvető helyesírási hibákon és választ adjon, de előfordulhat, hogy feltételezi, ez szándékos, és befolyásolja a válaszát.

**Ellenőrizd a beállításokat.** A temperature és a top_p beállítások szabályozzák, mennyire determinisztikus a modell a válasz generálásában. Ha olyan választ kérsz, ahol csak egy helyes megoldás van, ezeket alacsonyabbra érdemes állítani. Ha viszont sokféle válaszra van szükséged, akkor magasabbra. Az egyik leggyakoribb hiba, hogy ezeket „ügyesség” vagy „kreativitás” szabályozóként értelmezik.


Forrás: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Beküldés!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Szöveg összefoglalása  
#### Kihívás  
Összefoglalja a szöveget úgy, hogy a szövegrészlet végére egy 'tl;dr:' kifejezést ad hozzá. Figyeld meg, hogyan érti meg a modell, hogy számos feladatot el tud látni további utasítások nélkül. Kísérletezhetsz részletesebb promptokkal, mint a tl;dr, hogy módosítsd a modell viselkedését és személyre szabd a kapott összefoglalást(3).  

A legutóbbi kutatások jelentős előrelépéseket mutattak számos NLP feladaton és mérőszámon, amikor egy nagy szövegkorpuszra történő előzetes tanítás után egy adott feladatra finomhangolást végeztek. Bár az architektúra általában feladattól független, ez a módszer mégis igényli az adott feladathoz tartozó finomhangolási adatbázisokat, amelyek több ezer vagy akár tízezernyi példát tartalmaznak. Ezzel szemben az emberek általában csak néhány példából vagy egyszerű utasításokból is el tudnak végezni egy új nyelvi feladatot – ez pedig valami, amiben a jelenlegi NLP rendszerek még nagyrészt küzdenek. Itt bemutatjuk, hogy a nyelvi modellek méretezésének növelése jelentősen javítja a feladattól független, kevés példás teljesítményt, néha akár az előzőleg legjobb finomhangolási megközelítésekkel vetekedő eredményt is elérve.
