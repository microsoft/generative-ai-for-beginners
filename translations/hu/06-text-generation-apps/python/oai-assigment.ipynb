## Mi az a szövegalkotó alkalmazás?

Általában, amikor alkalmazást építesz, van valamilyen felülete, például:

- Parancs alapú. A konzol alkalmazások tipikus alkalmazások, ahol beírsz egy parancsot és az végrehajt egy feladatot. Például a `git` egy parancs alapú alkalmazás.
- Felhasználói felület (UI). Néhány alkalmazás grafikus felhasználói felülettel (GUI) rendelkezik, ahol gombokat kattintasz, szöveget adsz meg, opciókat választasz és így tovább.

### Konzol és UI alkalmazások korlátai

Összehasonlítva egy parancs alapú alkalmazással, ahol beírsz egy parancsot:

- **Korlátozott**. Nem írhatsz be tetszőleges parancsokat, csak azokat, amelyeket az alkalmazás támogat.
- **Nyelv-specifikus**. Néhány alkalmazás sok nyelvet támogat, de alapértelmezettként az alkalmazás egy konkrét nyelvre épül, még ha hozzá is adhatsz több nyelvi támogatást.

### A szövegalkotó alkalmazások előnyei

Miben különbözik tehát egy szövegalkotó alkalmazás?

Egy szövegalkotó alkalmazásban nagyobb rugalmassággal rendelkezel, nem vagy korlátozva egy sor parancsra vagy egy adott beviteli nyelvre. Ehelyett természetes nyelvet használhatsz az alkalmazással való interakcióhoz. Egy másik előny, hogy már egy olyan adattal dolgozol, amelyet hatalmas információgyűjteményen képeztek ki, míg egy hagyományos alkalmazás általában egy adatbázisra van korlátozva.

### Mit építhetek egy szövegalkotó alkalmazással?

Sok mindent építhetsz, például:

- **Chatbot**. Egy kérdésekre válaszoló chatbot témákról, például a cégedről és termékeiről tökéletes lehet.
- **Segítő**. Az LLM-ek (nagyméretű nyelvi modellek) remekek szövegek összefoglalására, szövegből való betekintések nyerésére, például önéletrajzok és hasonlók létrehozására.
- **Kódasszisztens**. A használt nyelvi modelltől függően építhetsz egy kódasszisztenst, amely segít a kódírásban. Például használhatsz olyan termékeket, mint a GitHub Copilot vagy a ChatGPT, hogy segítsenek a kód írásában.

## Hogyan kezdjek hozzá?

Nos, találnod kell egy módot az LLM-mel való integrációra, amely általában a következő két megközelítést jelenti:

- API használata. Itt webkéréseket építesz a promptoddal és visszakapod a generált szöveget.
- Könyvtár használata. A könyvtárak segítenek az API hívások csomagolásában és könnyebbé teszik a használatot.

## Könyvtárak/SDK-k

Számos ismert könyvtár létezik az LLM-ekkel való munkához, például:

- **openai**, ez a könyvtár megkönnyíti a modellhez való kapcsolódást és promptok küldését.

Vannak magasabb szintű könyvtárak is, például:

- **Langchain**. A Langchain jól ismert és támogatja a Python-t.
- **Semantic Kernel**. A Semantic Kernel egy Microsoft által fejlesztett könyvtár, amely támogatja a C#, Python és Java nyelveket.

## Szöveg generálása

A szöveg generálásának módja a `responses.create` metódus használata. Íme egy példa:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

A fenti kódban létrehozunk egy válasz objektumot, megadjuk a használni kívánt modellt és a promptot. Ezután kiírjuk a generált szöveget.

### Csevegési válaszok

Eddig láttad, hogyan generálhatunk szöveget egyetlen promptból. A Responses API jól használható chatbotokhoz is - egy üzenetek listáját adod meg bemenetként annak érdekében, hogy kialakíts egy beszélgetést. Íme egy példa a használatára:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Erről a funkcióról a későbbi fejezetben olvashatsz bővebben.

## Gyakorlat - készítsünk egy receptgenerátort

Most, hogy lejátszottuk a szcenáriót, írjuk meg a kódot a bemutatott szcenárióhoz illeszkedve. Ehhez kövesd az alábbi lépéseket:

1. Használd az meglévő *app.py* fájlt kiindulópontként
1. Keresd meg a `prompt` változót, és változtasd meg a kódját a következőre:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Nagyszerű, nézzük meg, hogyan javíthatunk a dolgokon. A javításhoz azt szeretnénk, hogy a kód rugalmas legyen, így az összetevők és a receptek száma is továbbfejleszthető és módosítható legyen.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
