## Mi az a szövegalkotó alkalmazás?

Általában, amikor egy alkalmazást építesz, az valamilyen interfészsel rendelkezik, például a következőképpen:

- Parancs-alapú. A konzolos alkalmazások tipikus példák, ahol beírsz egy parancsot, és az végrehajt egy feladatot. Például a `git` egy parancs-alapú alkalmazás.
- Felhasználói felület (UI). Néhány alkalmazás grafikus felhasználói felülettel (GUI) rendelkezik, ahol gombokra kattintasz, szöveget viszel be, választási lehetőségeket jelölsz stb.

### A konzolos és UI alkalmazások korlátai

Hasonlítsd össze egy parancs-alapú alkalmazással, ahol beírsz egy parancsot:

- **Korlátozott**. Nem írhatsz be bármilyen parancsot, csak azokat, amiket az alkalmazás támogat.
- **Nyelvspecifikus**. Néhány alkalmazás több nyelvet támogat, de alapértelmezetten az alkalmazás egy adott nyelvre épül, még ha hozzá is adhatók további nyelvtámogatások.

### A szövegalkotó alkalmazások előnyei

Miben más egy szövegalkotó alkalmazás?

Egy szövegalkotó alkalmazásban nagyobb a rugalmasság, nem vagy korlátozva egy adott parancssorra vagy egy adott bemenet nyelvére. Ehelyett természetes nyelvet használhatsz az alkalmazással való interakcióhoz. Egy másik előny, hogy egy már egy hatalmas szöveggyűjteményen kiképzett adatforrással dolgozol, míg egy hagyományos alkalmazás általában egy adatbázishoz van korlátozva.

### Mit építhetek egy szövegalkotó alkalmazással?

Sok mindenre használhatod. Például:

- **Chatbot**. Egy chatbot, ami válaszol kérdésekre különböző témákról, például a cégedről és termékeiről, jó választás lehet.
- **Segéd**. A nagynyelvű modellek (LLM-ek) kiválóak például szövegek összefoglalására, szöveg elemzésére, szöveg előállítására, mint például önéletrajzok írása és még sok másra.
- **Kódasszisztens**. Az általad használt nyelvi modelltől függően készíthetsz egy kódasszisztenst, amely segít kódot írni. Például használhatsz olyan termékeket, mint a GitHub Copilot vagy a ChatGPT, hogy segítsenek kódot írni.

## Hogyan kezdjek hozzá?

Nos, találnod kell egy módot arra, hogy integrálódj egy nagynyelvű modellhez (LLM), ami általában a következő két megközelítést jelenti:

- Használj API-t. Itt web kérésokat építesz a promptoddal, és visszakapod a generált szöveget.
- Használj könyvtárat. A könyvtárak segítenek az API hívások becsomagolásában, és könnyebbé teszik azok használatát.

## Könyvtárak/SDK-k

Van néhány ismert könyvtár, amelyeket nagynyelvű modellekkel való munkára használnak, például:

- **openai**, ez a könyvtár megkönnyíti a modellhez való csatlakozást és a promptok küldését.

Aztán vannak magasabb szintű könyvtárak, mint például:

- **Langchain**. A Langchain jól ismert és támogatja a Pythont.
- **Semantic Kernel**. A Semantic Kernel egy Microsoft könyvtár, amely támogatja a C#, a Python és a Java nyelveket.

## Szöveg generálása

A szöveg generálásához használd a `responses.create` metódust. Íme egy példa:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

A fenti kódban létrehozunk egy response objektumot, átadjuk a használni kívánt modellt és a promptot. Majd kiírjuk a generált szöveget.

### Chat válaszok

Eddig láttad, hogyan generálhatunk szöveget egyetlen promptból. A Responses API kiválóan alkalmas chatbotokhoz is - egy üzenetlistát adsz meg bemenetként (`input`), hogy felépüljön a beszélgetés. Íme egy példa a használatára:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)

```

Erről a funkcióról bővebben egy későbbi fejezetben.

## Gyakorlat - receptgenerátor készítése

Most, hogy lejátszottuk a forgatókönyvet, írjunk kódot a bemutatott esethez igazodva. Ehhez kövessük az alábbi lépéseket:

1. Használjuk az meglévő *app.py* fájlt kiindulópontként
1. Keressük meg a `prompt` változót, és cseréljük le a kódját az alábbiakra:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Itt azt javasoljuk, hogy adjunk hozzá egy további promptot, de ahhoz, hogy ez működjön, az előző prompt eredményét kontextusként kell átadni a következő promptnak.

    Keresd meg azt a részt a kódban, ahol az első prompt eredményét kiírja, és add hozzá az alábbi kódot alatta:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # válasz nyomtatása
    print("Shopping list:")
    print(response.output_text)
    ```

    Figyeld meg a következőket:

    - Új promptot építünk úgy, hogy az első prompt eredményét hozzáadjuk az új prompthoz:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Új kérést teszünk, de figyelembe véve az első promptban kért tokenek számát, így most a `max_output_tokens` értéke 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Ezt a kódot futtatva eljutunk a következő eredményhez:

        ```output

        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        ```

- **Egy szó a token hosszúságról**. Fontolóra kell vennünk, hogy hány tokenre van szükségünk a kívánt szöveg generálásához. A tokenek pénzbe kerülnek, így ahol lehet, gazdaságosan kell használnunk őket. Például, tudjuk-e úgy megfogalmazni a promptot, hogy kevesebb tokent használjunk?

   A használt tokenek számának megváltoztatásához használhatod a `max_output_tokens` paramétert. Például, ha 100 tokent szeretnél használni, így tennéd:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Kísérletezés a hőmérséklettel**. A hőmérsékletről eddig nem beszéltünk, de fontos kontextus arra, hogyan működik a programunk. Minél magasabb a hőmérsékleti érték, annál véletlenszerűbb lesz a kimenet. Ezzel szemben minél alacsonyabb a hőmérséklet, annál kiszámíthatóbb lesz a kimenet. Fontold meg, szeretnél-e változatosságot a kimenetedben vagy sem.

   A hőmérséklet megváltoztatásához használhatod a `temperature` paramétert. Például, ha 0.5-ös hőmérsékletet szeretnél, így tennéd:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Megjegyzés, minél közelebb van 1.0-hez, annál változatosabb lesz a kimenet.
