# Szövegalkotó alkalmazások készítése

Eddig a tananyag során láttad, hogy vannak alapvető fogalmak, mint például a promptok, és létezik egy egész diszciplína, amit "prompt mérnökségnek" hívnak. Sok eszköz, amellyel interakcióba léphetsz, például a ChatGPT, az Office 365, a Microsoft Power Platform és még sok más, támogatja azt, hogy promptokat használj valaminek az eléréséhez.

Ahhoz, hogy ilyen élményt adj egy alkalmazáshoz, meg kell értened olyan fogalmakat, mint a promptok, a kiegészítések, és ki kell választanod egy könyvtárat a munkához. Pont ezt tanulod meg ebben a fejezetben.

## Mi az a szövegalkotó alkalmazás?

Általában, amikor egy alkalmazást építesz, valamilyen felülete van, például az alábbiak szerint:

- Parancs-alapú. A konzolalkalmazások tipikus példái azoknak az appoknak, ahol bepötyögsz egy parancsot, és az végrehajt egy feladatot. Például a `git` egy parancs-alapú alkalmazás.
- Felhasználói felület (UI). Néhány alkalmazás grafikus felhasználói felülettel (GUI) rendelkezik, ahol gombokat kattintasz, szöveget viszel be, opciókat választasz, és még sok más.

### A konzol- és UI alkalmazások korlátozottak

Hasonlítsd össze egy parancs-alapú alkalmazással, ahol beírsz egy parancsot:

- **Korlátozott**. Nem irkálhatsz be tetszőleges parancsot, csak azokat, amelyeket az app támogat.
- **Nyelv specifikus**. Egyes alkalmazások több nyelvet támogatnak, de alapból az alkalmazást egy konkrét nyelvhez építik fel, még ha további nyelvi támogatást is hozzáadhatsz.

### A szövegalkotó alkalmazások előnyei

Miben más egy szövegalkotó alkalmazás?

Egy szövegalkotó alkalmazásban több a rugalmasság, nem vagy korlátozva egy adott parancskészletre vagy specifikus bemeneti nyelvre. Ehelyett természetes nyelvet használhatsz az alkalmazással való interakcióhoz. Egy további előny az, hogy már egy olyan adatforrással kommunikálsz, amelyet egy hatalmas információs korpuszon képeztek, míg egy hagyományos alkalmazás esetleg korlátozva van a benne lévő adatbázis tartalmára.

### Mit építhetek szövegalkotó alkalmazással?

Sok mindent építhetsz. Például:

- **Egy chatbotot**. Egy chatbot, amely válaszol kérdésekre a témákról, például a cégedről és termékeiről, jól illik ehhez.
- **Segítőt**. A nagy nyelvi modellek (LLM-ek) kiválóak olyan feladatokban, mint a szövegek összefoglalása, a szövegből való betekintés nyerése, szövegek előállítása, például önéletrajzok és még sok más.
- **Kódtámogatót**. A használt nyelvi modelltől függően készíthetsz kódtámogatót, ami segít kódot írni. Például használhatod a GitHub Copilot terméket, valamint a ChatGPT-t is a kódírás segítéséhez.

## Gyakorlat - recept generátor készítése

Most, hogy lejátszottunk egy forgatókönyvet, írjunk kódot, ami megfelel a bemutatott forgatókönyvnek. Ehhez kövesd az alábbi lépéseket:

1. Használd a meglévő fájlt kiindulópontként
1. Hozz létre egy `prompt` változót, és módosítsd az alábbiak szerint a mintakódot:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Nagyszerű, nézzük meg, hogyan javíthatunk a dolgokon. Ahhoz, hogy javítsunk a dolgokon, biztosítani akarjuk, hogy a kód rugalmas legyen, így az összetevők és a receptek száma fejleszthető és módosítható legyen.


1. Változtassuk meg a kódot a következő módon:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

    Ehhez a funkcióhoz vagy mindent egy promptban próbálhatunk megoldani, vagy két részre bonthatjuk. Próbáljuk meg az utóbbit. Itt egy további prompt hozzáadását javasoljuk, de ahhoz, hogy ez működjön, az első prompt eredményét kontextusként kell adnunk a második promptnak.

    Keresse meg a kódban azt a részt, amely kiírja az első prompt eredményét, és adja hozzá a következő kódot alá:

    old_prompt_result = response.choices[0].message.content

        messages=[
            {

- **Kísérletezés a hőmérséklettel**. A temperature (hőmérséklet) paraméterről eddig nem esett szó, de fontos a program teljesítményének szempontjából. Minél magasabb az értéke, annál véletlenszerűbb lesz a kimenet. Ezzel ellentétben minél alacsonyabb az érték, annál kiszámíthatóbb a válasz. Fontolja meg, szeretne-e változatosságot a kimenetben.
