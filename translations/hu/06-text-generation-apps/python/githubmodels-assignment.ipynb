## Mi az a szöveggeneráló alkalmazás?

Általában amikor egy alkalmazást építesz, van valamilyen felülete, például az alábbiak szerint:

- Parancs-alapú. Konzolos alkalmazások tipikus példák, ahol egy parancsot írsz be, és az végrehajt egy feladatot. Például a `git` parancs-alapú alkalmazás.
- Felhasználói felület (UI). Néhány alkalmazás grafikus felhasználói felülettel rendelkezik (GUI), ahol kattintasz gombokra, szöveget írsz be, opciókat választasz és így tovább.

### A konzolos és UI alkalmazások korlátozottak

Hasonlítsd össze egy parancs-alapú alkalmazással, ahol parancsot írsz:

- **Korlátozottak.** Nem írhatsz csak úgy bármilyen parancsot, csak azokat, amelyeket az alkalmazás támogat.
- **Nyelvspecifikusak.** Néhány alkalmazás sok nyelvet támogat, de alapértelmezetten az alkalmazás egy adott nyelvhez készült, még ha többnyelvű támogatást is adhatsz hozzá.

### A szöveggeneráló alkalmazások előnyei

Miben más egy szöveggeneráló alkalmazás?

Egy szöveggeneráló alkalmazásban nagyobb a rugalmasság, nem vagy korlátozva parancsokra vagy adott bemeneti nyelvre. A természetes nyelvet használhatod az alkalmazással való interakcióhoz. Egy további előny, hogy már egy hatalmas információhalmazon kiképzett adatforrással kommunikálsz, míg egy hagyományos alkalmazás adatbázisra korlátozódhat.

### Mit építhetek egy szöveggeneráló alkalmazással?

Sok mindent építhetsz. Például:

- **Chatbotot.** Egy chatbot, amely kérdésekre válaszol témákról, például a cégedről és termékeiről, jól jöhet.
- **Segédeszközt.** A nagy nyelvi modellek jól összefoglalnak szöveget, szövegből nyernek ki adatokat, készítenek önéletrajzokat és még sok mást.
- **Kódtámogatót.** A használt nyelvi modell függvényében építhetsz kódtámogatót, amely segít a kódírásban. Például használhatsz termékeket, mint a GitHub Copilot, illetve ChatGPT-t is a kódírás segítésére.

## Feladat - receptgenerátor készítése

Most, hogy eljátszottuk a szcenáriót, írjuk meg a kódot a bemutatott szcenárióhoz. Ehhez kövesd az alábbi lépéseket:

1. Használd a meglévő fájlt kiindulási pontként
1. Hozz létre egy `prompt` változót, és módosítsd a mintakódot az alábbiak szerint:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Remek, nézzük meg, hogyan javíthatunk a dolgokon. A fejlesztés érdekében azt szeretnénk, hogy a kód rugalmas legyen, így az összetevők és a receptek száma javítható és módosítható legyen.


1. Változtassuk meg a kódot a következő módon:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
