## Generiranje teksta

Način generiranja teksta je korištenjem metode `responses.create`. Evo primjera:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

U gore navedenom kodu stvaramo objekt odgovora i prosljeđujemo model koji želimo koristiti i prompt. Zatim ispisujemo generirani tekst.

### Odgovori za chat

Do sada ste vidjeli kako generirati tekst iz jednog prompta. Responses API je također vrlo pogodan za chatbote - dodajete popis poruka kao `input` za izgradnju razgovora. Evo primjera korištenja:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Više o ovoj funkcionalnosti u nadolazećem poglavlju.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    Za ovu funkcionalnost mogli bismo pokušati sve riješiti u jednom promptu ili podijeliti u dva prompta. Pokušajmo s drugim pristupom. Ovdje predlažemo dodavanje dodatnog prompta, ali da bi to funkcioniralo, potrebno je dodati rezultat prvog prompta kao kontekst drugom promptu. 

    Pronađite dio koda koji ispisuje rezultat prvog prompta i dodajte sljedeći kod ispod:
    
    ```python

    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # ispiši odgovor
    print("Shopping list:")
    print(response.output_text)
    ```

    Obratite pažnju na sljedeće:

    - Kreiramo novi prompt dodajući rezultat prvog prompta u novi prompt: 
    
        ```python

        messages = [{"role": "user", "content": new_prompt}]
        ```

- **Riječ o duljini tokena**. Trebali bismo razmotriti koliko tokena nam treba za generiranje željenog teksta. Tokeni koštaju novac, pa bismo, gdje je moguće, trebali biti štedljivi s brojem korištenih tokena. Na primjer, možemo li formulirati prompt tako da koristimo manje tokena?

- **Eksperimentiranje s temperaturom**. Temperatura je nešto o čemu dosad nismo govorili, ali je važan kontekst za rad naše aplikacije. Što je viša vrijednost temperature, izlaz će biti nasumičniji.
