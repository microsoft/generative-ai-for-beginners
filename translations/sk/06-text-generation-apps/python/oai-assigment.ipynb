## Čo je aplikácia na generovanie textu?

Zvyčajne, keď vytvárate aplikáciu, má nejaké rozhranie ako napríklad:

- Príkazové rozhranie. Konzolové aplikácie sú typické aplikácie, kde napíšete príkaz a vykoná sa úloha. Napríklad `git` je aplikácia založená na príkazoch.
- Užívateľské rozhranie (UI). Niektoré aplikácie majú grafické užívateľské rozhranie (GUI), kde klikáte na tlačidlá, vkladáte text, vyberáte možnosti a podobne.

### Konzolové a UI aplikácie sú obmedzené

Porovnajte to s príkazovou aplikáciou, kde píšete príkaz:

- **Je to obmedzené**. Nemôžete napísať akýkoľvek príkaz, iba tie, ktoré aplikácia podporuje.
- **Špecifické na jazyk**. Niektoré aplikácie podporujú viacero jazykov, ale štandardne sú vytvorené pre určitý jazyk, aj keď môžete pridať podporu ďalších jazykov.

### Výhody aplikácií na generovanie textu

Ako sa teda aplikácia na generovanie textu líši?

V aplikácii na generovanie textu máte väčšiu flexibilitu, nie ste obmedzení na súbor príkazov alebo konkrétny vstupný jazyk. Namiesto toho môžete komunikovať prirodzeným jazykom. Ďalšou výhodou je, že pracujete s dátovým zdrojom, ktorý bol trénovaný na obsiahlej množine informácií, zatiaľ čo tradičná aplikácia môže byť obmedzená na dáta v databáze.

### Čo môžem vytvoriť s aplikáciou na generovanie textu?

Môžete vytvoriť mnoho vecí, napríklad:

- **Chatbot**. Chatbot odpovedajúci na otázky o témach, ako je vaša firma a jej produkty, môže byť dobrým riešením.
- **Asistent**. Veľké jazykové modely (LLM) sú skvelé na veci ako sumarizácia textu, získavanie poznatkov z textu, tvorbu textov ako životopisov a ďalšie.
- **Asistent pre kódovanie**. Podľa použitého jazykového modelu môžete vytvoriť asistenta, ktorý vám pomáha písať kód. Napríklad môžete použiť produkty ako GitHub Copilot alebo ChatGPT na pomoc s písaním kódu.

## Generovanie textu

Na generovanie textu použite metódu `responses.create`. Tu je príklad:

```python

```

V uvedenom kóde vytvoríme objekt odpovede a zadáme model, ktorý chceme použiť, a prompt. Potom vypíšeme vygenerovaný text.

### Odpovede chatu

Doteraz ste videli, ako generovať text z jedného promptu. API Responses je však dobre použiteľné aj pre chatboty – poskytnete zoznam správ ako `input` na vytvorenie konverzácie. Tu je príklad použitia:

```python

  )
```

- **Vytvoriť nákupný zoznam**. Chceme vytvoriť nákupný zoznam, ktorý zohľadní, čo už doma máme.

    Pre túto funkciu môžeme buď skúsiť vyriešiť všetko v jednom promptu, alebo to môžeme rozdeliť na dva prompty. Skúsme druhý prístup. Navrhujeme pridať ďalší prompt, ale aby to fungovalo, potrebujeme pridať výsledok prvého promptu ako kontext k druhému promptu.

    Nájdite časť v kóde, ktorá vypisuje výsledok z prvého promptu a pridajte nasledujúci kód pod ňu:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # vytlačiť odpoveď
    print("Shopping list:")
    print(response.output_text)
    ```

    Dajte pozor na nasledovné:

    - Skladáme nový prompt tak, že pridávame výsledok z prvého promptu do nového promptu:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Vykonávame nový request, pričom zohľadňujeme počet tokenov, o ktoré sme žiadali v prvom promte, toto číslo nastavujeme na `max_output_tokens` 1200.

