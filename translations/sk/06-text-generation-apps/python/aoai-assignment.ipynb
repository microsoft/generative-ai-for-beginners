## Čo je aplikácia na generovanie textu?

Normálne, keď vytvárate aplikáciu, má nejaké rozhranie ako napríklad toto:

- Príkazové rozhranie. Konzolové aplikácie sú typické aplikácie, kde zadáte príkaz a ten vykoná úlohu. Napríklad `git` je príkazová aplikácia.
- Používateľské rozhranie (UI). Niektoré aplikácie majú grafické používateľské rozhranie (GUI), kde klikáte na tlačidlá, zadávate text, vyberáte možnosti a viac.

### Konzolové a UI aplikácie sú obmedzené

Porovnajte to s príkazovou aplikáciou, kde zadávate príkaz:

- **Je to obmedzené**. Nemôžete zadať ľubovoľný príkaz, iba tie, ktoré aplikácia podporuje.
- **Špecifické pre jazyk**. Niektoré aplikácie podporujú viacero jazykov, ale predvolene je aplikácia vytvorená pre konkrétny jazyk, aj keď môžete pridať podporu ďalších jazykov.

### Výhody aplikácií na generovanie textu

Ako sa teda líši aplikácia na generovanie textu?

V aplikácii na generovanie textu máte väčšiu flexibilitu, nie ste obmedzení na sadu príkazov alebo konkrétny vstupný jazyk. Namiesto toho môžete používať prirodzený jazyk na interakciu s aplikáciou. Ďalšou výhodou je, že už interagujete so zdrojom dát, ktorý bol vycvičený na obrovskom korpuse informácií, zatiaľ čo tradičná aplikácia môže byť obmedzená na to, čo je v databáze.

### Čo môžem vytvoriť pomocou aplikácie na generovanie textu?

Môžete vytvoriť mnoho vecí. Napríklad:

- **Chatbota**. Chatbot, ktorý odpovedá na otázky o témach ako vaša spoločnosť a jej produkty, by mohol byť dobrým riešením.
- **Pomocníka**. Veľké jazykové modely sú skvelé v veciach ako zhrnutie textu, získavanie poznatkov z textu, tvorba textu ako životopisy a viac.
- **Asistenta kódu**. Podľa použitého jazykového modelu môžete vytvoriť asistenta kódu, ktorý vám pomáha písať kód. Napríklad môžete použiť produkt ako GitHub Copilot alebo ChatGPT, ktorý vám pomôže písať kód.

## Generovanie textu

Spôsob, ako generovať text, je použiť metódu `responses.create`. Tu je príklad:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

V uvedenom kóde vytvoríme objekt odpovede a zadáme model, ktorý chceme použiť, a prompt. Potom vytlačíme vygenerovaný text.

### Chat odpovede
