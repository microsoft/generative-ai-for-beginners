# Úvod do návrhu promptov
Návrh promptov je proces navrhovania a optimalizácie promptov pre úlohy spracovania prirodzeného jazyka. Zahŕňa výber správnych promptov, ladenie ich parametrov a hodnotenie ich výkonu. Návrh promptov je kľúčový pre dosiahnutie vysokej presnosti a efektívnosti v NLP modeloch. V tejto sekcii preskúmame základy návrhu promptov s využitím modelov OpenAI na prieskum.


### Cvičenie 1: Tokenizácia
Preskúmajte tokenizáciu pomocou tiktoken, rýchleho open-source tokenizéra od OpenAI
Pozrite si [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) pre viac príkladov.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Cvičenie 2: Overenie nastavenia kľúča API OpenAI

Spustite nižšie uvedený kód, aby ste overili, či je váš koncový bod OpenAI správne nastavený. Kód sa len pokúša o jednoduchý základný príkaz a overuje dokončenie. Vstup `oh say can you see` by sa mal dokončiť niečím v zmysle `by the dawn's early light..`


````python
# Uses the OpenAI client with the Responses API.
# See https://platform.openai.com/docs/api-reference/responses

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### Cvičenie 4: Na základe inštrukcie 
Použite premennú "text" na nastavenie primárneho obsahu 
a premennú "prompt" na poskytnutie inštrukcie súvisiacej s týmto primárnym obsahom.

Tu žiadame model, aby zhrnul text pre žiaka druhého ročníka


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### Cvičenie 5: Komplexný prompt 
Vyskúšajte požiadavku, ktorá obsahuje systémové, používateľské a asistenčné správy 
Systém nastavuje kontext asistenta
Používateľské a asistenčné správy poskytujú kontext viackolovej konverzácie

Všimnite si, ako je osobnosť asistenta nastavená na "sarkastickú" v systémovom kontexte. 
Vyskúšajte použiť inú osobnostnú konfiguráciu. Alebo skúste inú sériu vstupno-výstupných správ


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
