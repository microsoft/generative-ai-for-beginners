## برنامه تولید متن چیست؟

معمولاً وقتی برنامه‌ای می‌سازید، یک نوع رابط کاربری مثل موارد زیر دارد:

- مبتنی بر فرمان. برنامه‌های کنسول معمولاً برنامه‌هایی هستند که شما یک فرمان می‌نویسید و آن یک وظیفه را انجام می‌دهد. مثلاً، `git` یک برنامه مبتنی بر فرمان است.
- رابط کاربری (UI). بعضی برنامه‌ها رابط‌های گرافیکی کاربری (GUI) دارند که در آن دکمه‌ها را کلیک می‌کنید، متن وارد می‌کنید، گزینه‌ها را انتخاب می‌کنید و غیره.

### برنامه‌های کنسول و رابط کاربری محدودیت دارند

آن را با یک برنامه مبتنی بر فرمان مقایسه کنید که فرمانی تایپ می‌کنید:

- **محدود است**. نمی‌توانید هر فرمانی را تایپ کنید، فقط آن‌هایی که برنامه پشتیبانی می‌کند.
- **زبان خاص**. بعضی برنامه‌ها زبان‌های زیادی را پشتیبانی می‌کنند، اما به طور پیش‌فرض برنامه برای یک زبان خاص ساخته شده است، حتی اگر بتوانید پشتیبانی زبان بیشتری اضافه کنید.

### مزایای برنامه‌های تولید متن

پس برنامه تولید متن چه تفاوتی دارد؟

در یک برنامه تولید متن، انعطاف بیشتری دارید، محدود به مجموعه‌ای از فرمان‌ها یا زبان ورودی خاص نیستید. در عوض، می‌توانید از زبان طبیعی برای تعامل با برنامه استفاده کنید. مزیت دیگر این است که چون شما در حال کار با منبع داده‌ای هستید که بر روی یک مجموعه عظیم از اطلاعات آموزش دیده، در حالی که یک برنامه سنتی ممکن است فقط به چیزی که در یک پایگاه داده است محدود باشد.

### با برنامه تولید متن چه می‌توانم بسازم؟

چیزهای زیادی می‌توانید بسازید. به عنوان مثال:

- **چت‌بات**. یک چت‌بات که به سوالات درباره موضوعات مختلف، مثل شرکت شما و محصولاتش پاسخ می‌دهد، می‌تواند گزینه خوبی باشد.
- **دستیار**. مدل‌های زبان بزرگ در چیزهایی مثل خلاصه کردن متن، استخراج بینش از متن، تولید متن مثل رزومه و بیشتر، عالی هستند.
- **دستیار کدنویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید دستیار کدی بسازید که به نوشتن کد کمک کند. به عنوان مثال، می‌توانید از محصولی مثل GitHub Copilot و همچنین ChatGPT برای کمک به نوشتن کد استفاده کنید.

- استفاده از کتابخانه. کتابخانه‌ها به صورت درونی تماس‌های API را پوشش می‌دهند و استفاده از آن‌ها را آسان‌تر می‌کنند.

## تولید متن

راه تولید متن استفاده از متد `responses.create` است. در اینجا یک مثال است:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

در کد بالا، یک شی پاسخ ایجاد می‌کنیم و مدل و پرامپتی که می‌خواهیم استفاده کنیم را می‌فرستیم. سپس متن تولید شده را چاپ می‌کنیم.

### پاسخ‌های چت

تاکنون دیده‌اید چگونه می‌توان از یک پرامپت منفرد متن تولید کرد. API پاسخ‌ها همچنین برای چت‌بات‌ها بسیار مناسب است - شما فهرستی از پیام‌ها را به عنوان `input` برای ایجاد یک گفتگو ارائه می‌دهید. اینجا یک مثال از استفاده آن است:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

در مورد این قابلیت در فصل‌های بعد بیشتر خواهید آموخت.

## تمرین - اولین برنامه تولید متن شما

حالا که یاد گرفتید چگونه سرویس OpenAI را راه‌اندازی و پیکربندی کنید، زمان ساخت اولین برنامه تولید متن شماست. برای ساخت برنامه خود، این مراحل را دنبال کنید:


1. یک محیط مجازی ایجاد کنید و openai را نصب کنید:

  > [!NOTE] اگر این دفترچه یادداشت را در Codespaces یا درون یک Devcontainer اجرا کنید، این مرحله لازم نیست


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> اگر از ویندوز استفاده می‌کنید، به جای `source venv/bin/activate` دستور `venv\Scripts\activate` را وارد کنید.

> [!NOTE]
> کلید OpenAI خود را با رفتن به https://platform.openai.com/settings/organization/api-keys پیدا کنید و به دنبال `API keys` بگردید. شما می‌توانید یک کلید جدید در آنجا ایجاد کرده و بلافاصله مقدار آن را کپی کنید.


1. یک فایل *app.py* ایجاد کنید و کد زیر را در آن قرار دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    شما باید خروجی شبیه به زیر ببینید:

    ```output

    روزی روزگاری یک پری دریایی بسیار ناراضی وجود داشت.
    ```


## انواع مختلف پرامپت‌ها، برای کارهای مختلف

## تمرین - ساخت یک تولیدکننده دستور پخت

حال که یک سناریو را اجرا کردیم، بیایید کدی بنویسیم که با سناریوی نمایش‌داده‌شده مطابقت داشته باشد. برای این کار، مراحل زیر را دنبال کنید:

1. از فایل *app.py* موجود به عنوان نقطه شروع استفاده کنید
1. متغیر `prompt` را پیدا کنید و کد آن را به صورت زیر تغییر دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

عالی است، بیایید ببینیم چگونه می‌توانیم اوضاع را بهتر کنیم. برای بهبود اوضاع، می‌خواهیم مطمئن شویم کد انعطاف‌پذیر است، بنابراین مواد اولیه و تعداد دستورها می‌توانند بهبود یافته و تغییر کنند.


1. بیاید کد را به شکل زیر تغییر دهیم:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

اجرای کد برای یک آزمایش می‌تواند به این شکل باشد:
    
```output

List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

    برای این قابلیت، می‌توانیم همه را در یک پرامپت حل کنیم یا آن را به دو پرامپت تقسیم کنیم. رویکرد دوم را امتحان کنیم. در اینجا پیشنهاد می‌کنیم یک پرامپت اضافی اضافه کنیم، اما برای اینکه این کار عملی شود، باید نتیجه پرامپت اول را به عنوان زمینه به پرامپت دوم اضافه کنیم.

    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # پاسخ را چاپ کن
    print("Shopping list:")
    print(response.output_text)
    ```
