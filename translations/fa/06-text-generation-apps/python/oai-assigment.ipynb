## برنامه تولید متن چیست؟

معمولاً وقتی یک برنامه می‌سازید، یک نوع رابط کاربری دارد مانند موارد زیر:

- مبتنی بر دستور. برنامه‌های کنسولی نمونه‌هایی هستند که در آن‌ها شما یک دستور می‌نویسید و برنامه کاری را انجام می‌دهد. برای مثال، `git` یک برنامه مبتنی بر دستور است.
- رابط کاربری (UI). برخی برنامه‌ها رابط‌های گرافیکی کاربری دارند که در آن‌ها روی دکمه‌ها کلیک می‌کنید، متن وارد می‌کنید، گزینه انتخاب می‌کنید و غیره.

### برنامه‌های کنسولی و رابط کاربری محدودیت دارند

آن را با یک برنامه مبتنی بر دستور مقایسه کنید که در آن شما فرمان می‌دهید:

- **محدود است**. شما نمی‌توانید هر دستوری را بنویسید، فقط دستوراتی که برنامه پشتیبانی می‌کند.
- **خاص زبان است**. برخی برنامه‌ها از زبان‌های زیادی پشتیبانی می‌کنند، اما به طور پیش‌فرض برنامه برای یک زبان خاص ساخته شده است، حتی اگر بتوانید پشتیبانی از زبان‌های بیشتر را اضافه کنید.

### مزایای برنامه‌های تولید متن

پس یک برنامه تولید متن چگونه متفاوت است؟

در یک برنامه تولید متن، انعطاف‌پذیری بیشتری دارید، محدود به مجموعه‌ای از دستورات یا زبان ورودی خاص نیستید. در عوض، می‌توانید از زبان طبیعی برای تعامل با برنامه استفاده کنید. مزیت دیگر این است که در حال حاضر با یک منبع داده که بر روی حجم عظیمی از اطلاعات آموزش دیده در تعامل هستید، در حالی که یک برنامه سنتی ممکن است محدود به چیزی باشد که در یک پایگاه داده است.

### با برنامه تولید متن چه چیزهایی می‌توانم بسازم؟

چیزهای زیادی می‌توانید بسازید. برای مثال:

- **یک چت‌بات**. یک چت‌بات که به سوالاتی درباره موضوعات مختلف، مانند شرکت شما و محصولاتش پاسخ می‌دهد می‌تواند گزینه خوبی باشد.
- **دستیار**. مدل‌های زبان بزرگ (LLM) در کارهایی مانند خلاصه‌سازی متن، استخراج بینش از متن، تولید متنی مانند رزومه و غیره بسیار خوب هستند.
- **دستیار برنامه‌نویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید دستیار برنامه‌نویسی بسازید که به شما در نوشتن کد کمک کند. برای مثال، می‌توانید از محصولی مانند GitHub Copilot و همچنین ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چگونه شروع کنم؟

خب، باید راهی برای ارتباط با یک مدل زبان بزرگ (LLM) پیدا کنید که معمولاً دو رویکرد زیر را شامل می‌شود:

- استفاده از API. در این حالت، درخواست‌های وب را با پرامپت خود می‌سازید و متن تولید شده را دریافت می‌کنید.
- استفاده از یک کتابخانه. کتابخانه‌ها به کپسوله‌سازی تماس‌های API کمک می‌کنند و استفاده از آن‌ها را آسان‌تر می‌کنند.

## کتابخانه‌ها/SDKها

چند کتابخانه شناخته شده برای کار با مدل‌های زبان بزرگ وجود دارد مانند:

- **openai** که این کتابخانه اتصال به مدل شما و ارسال پرامپت‌ها را آسان می‌کند.

سپس کتابخانه‌هایی هستند که در سطح بالاتری عمل می‌کنند مانند:

- **Semantic Kernel** که کتابخانه‌ای از مایکروسافت است و از زبان‌های C#، Python و Java پشتیبانی می‌کند.

### تنظیم پیکربندی OpenAI

اگر از OpenAI استفاده می‌کنید، تنظیم پیکربندی به این صورت است:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

در بالا موارد زیر را تنظیم کرده‌ایم:

- `api_key`، این کلید API شما است که در داشبورد OpenAI یافت می‌شود.
- `deployment`، این نسخه GPT شما است.

> [!NOTE]
> `os.environ` تابعی است که متغیرهای محیطی را می‌خواند. می‌توانید از آن برای خواندن متغیرهای محیطی مانند `OPENAI_API_KEY` استفاده کنید.

## تولید متن

روش تولید متن استفاده از متد `responses.create` است. در اینجا یک مثال آمده است:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

در کد بالا، یک شیء پاسخ ایجاد می‌کنیم و مدلی را که می‌خواهیم استفاده کنیم و پرامپت را می‌دهیم. سپس متن تولید شده را چاپ می‌کنیم.

### پاسخ‌های چت

تا اینجا دیده‌اید چگونه می‌توان از یک پرامپت برای تولید متن استفاده کرد. API پاسخ‌ها همچنین برای چت‌بات‌ها بسیار مناسب است - یک فهرست پیام‌ها را به عنوان `input` ارائه می‌دهید تا یک مکالمه ساخته شود. در اینجا مثالی از استفاده آن آمده است:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

جزئیات بیشتر درباره این قابلیت در فصل‌های آینده ارائه خواهد شد.

## تمرین - اولین برنامه تولید متن شما

اکنون که یاد گرفته‌اید چگونه سرویس OpenAI را تنظیم و پیکربندی کنید، وقت آن است که اولین برنامه تولید متن خود را بسازید. برای ساخت برنامه خود، مراحل زیر را دنبال کنید:


1. یک محیط مجازی ایجاد کرده و openai را نصب کنید:

  > [!NOTE] اگر این نوت‌بوک را در Codespaces یا درون یک Devcontainer اجرا می‌کنید، این مرحله لازم نیست

> [!NOTE]

> [!NOTE]
> کلید OpenAI خود را با رفتن به https://platform.openai.com/settings/organization/api-keys  پیدا کنید و در بخش `API keys` جستجو کنید. می‌توانید یک کلید جدید بسازید و بلافاصله مقدار آن را کپی کنید.


1. یک فایل *app.py* بسازید و کد زیر را در آن قرار دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

    ```output

    ```


## انواع مختلف دستورات، برای کارهای مختلف

## تمرین - ساخت یک تولیدکننده دستور پخت

حال که یک سناریو را اجرا کرده‌ایم، بیایید کدی بنویسیم که با سناریوی نشان داده شده مطابقت داشته باشد. برای این کار، مراحل زیر را دنبال کنید:

1. از فایل موجود *app.py* به عنوان نقطه شروع استفاده کنید
1. متغیر `prompt` را پیدا کنید و کد آن را به صورت زیر تغییر دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

اگر اکنون کد را اجرا کنید، باید خروجی مشابهی ببینید:

عالی است، بیایید ببینیم چگونه می‌توانیم اوضاع را بهتر کنیم. برای بهتر شدن، می‌خواهیم مطمئن شویم کد انعطاف‌پذیر است، بنابراین مواد تشکیل‌دهنده و تعداد دستورهای آشپزی قابل بهبود و تغییر باشند.


1. بیایید کد را به شکل زیر تغییر دهیم:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

اجرای آزمایشی کد می‌تواند چیزی شبیه به این باشد:
    
```output

-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

    برای این قابلیت، می‌توانیم یا تمام کارها را در یک پرامپت انجام دهیم یا آن را به دو پرامپت تقسیم کنیم. حالت دوم را امتحان کنیم. در اینجا پیشنهاد می‌دهیم یک پرامپت اضافی اضافه کنیم، ولی برای آن باید نتیجه پرامپت اول را به پرامپت دوم به عنوان متن زمینه اضافه کنیم.

    بخشی از کد که نتیجه پرامپت اول را چاپ می‌کند پیدا کنید و کد زیر را زیر آن اضافه کنید:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # چاپ پاسخ
    print("Shopping list:")
    print(response.output_text)
    ```
