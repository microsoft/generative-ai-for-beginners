## اپلیکیشن تولید متن چیست؟

معمولاً وقتی یک اپلیکیشن می‌سازید، یک نوع رابط کاربری مثل موارد زیر دارد:

- مبتنی بر فرمان (Command-based). اپلیکیشن‌های کنسولی نمونه‌هایی هستند که در آن‌ها یک فرمان تایپ می‌کنید و یک کار انجام می‌شود. مثلاً `git` یک اپلیکیشن مبتنی بر فرمان است.
- رابط کاربری (UI). بعضی اپلیکیشن‌ها رابط کاربری گرافیکی (GUI) دارند که در آن‌ها دکمه کلیک می‌کنید، متن وارد می‌کنید، گزینه انتخاب می‌کنید و غیره.

### اپلیکیشن‌های کنسولی و UI محدودیت دارند

مقایسه کنید با یک اپلیکیشن مبتنی بر فرمان که در آن یک فرمان تایپ می‌کنید:

- **محدود است**. نمی‌توانید هر فرمانی را تایپ کنید، فقط فرمان‌هایی که اپلیکیشن پشتیبانی می‌کند.
- **وابسته به زبان**. بعضی اپلیکیشن‌ها از زبان‌های مختلف پشتیبانی می‌کنند، اما به طور پیش‌فرض برای یک زبان خاص ساخته شده‌اند، حتی اگر بتوانید پشتیبانی زبان‌های دیگر را اضافه کنید.

### مزایای اپلیکیشن‌های تولید متن

پس اپلیکیشن تولید متن چه تفاوتی دارد؟

در یک اپلیکیشن تولید متن، انعطاف بیشتری دارید، محدود به مجموعه‌ای از فرمان‌ها یا یک زبان ورودی خاص نیستید. در عوض می‌توانید با زبان طبیعی با اپلیکیشن تعامل کنید. مزیت دیگر این است که چون با یک منبع داده که روی حجم عظیمی از اطلاعات آموزش دیده کار می‌کنید، محدودیت‌های یک اپلیکیشن سنتی که فقط به دیتابیس وابسته است را ندارید.

### با اپلیکیشن تولید متن چه چیزهایی می‌توان ساخت؟

چیزهای زیادی می‌توانید بسازید. مثلاً:

- **چت‌بات**. یک چت‌بات که به سوالات درباره موضوعاتی مثل شرکت و محصولات شما پاسخ می‌دهد، می‌تواند گزینه خوبی باشد.
- **دستیار**. مدل‌های زبانی بزرگ (LLM) در کارهایی مثل خلاصه‌سازی متن، استخراج بینش از متن، تولید متن مثل رزومه و غیره عالی هستند.
- **دستیار کدنویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید یک دستیار کدنویسی بسازید که به شما در نوشتن کد کمک کند. مثلاً می‌توانید از محصولاتی مثل GitHub Copilot یا ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چطور می‌توانم شروع کنم؟

باید راهی برای ارتباط با یک مدل زبانی بزرگ (LLM) پیدا کنید که معمولاً شامل دو روش زیر است:

- استفاده از API. در این روش، درخواست‌های وب با پرامپت خود می‌سازید و متن تولیدشده را دریافت می‌کنید.
- استفاده از کتابخانه. کتابخانه‌ها فراخوانی‌های API را کپسوله می‌کنند و استفاده از آن‌ها را ساده‌تر می‌کنند.

## تولید متن

برای تولید متن باید از کلاس `chat.completion` استفاده کنید. مثال:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

در کد بالا، یک شی completion می‌سازیم و مدل مورد نظر و پرامپت را به آن می‌دهیم. سپس متن تولیدشده را چاپ می‌کنیم.

### Chat completions

تا اینجا دیدید که چطور از `Completion` برای تولید متن استفاده کردیم. اما یک کلاس دیگر به نام `ChatCompletion` هم وجود دارد که برای چت‌بات‌ها مناسب‌تر است. مثال استفاده از آن:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

در فصل بعدی بیشتر درباره این قابلیت صحبت خواهیم کرد.

## تمرین - اولین اپلیکیشن تولید متن شما

حالا که یاد گرفتیم چطور سرویس OpenAI را راه‌اندازی و پیکربندی کنیم، وقت آن است که اولین اپلیکیشن تولید متن خود را بسازید. برای ساخت اپلیکیشن خود، مراحل زیر را دنبال کنید:


۱. یک محیط مجازی بسازید و openai را نصب کنید:

  > [!NOTE] این مرحله لازم نیست اگر این نوت‌بوک را در Codespaces یا داخل یک Devcontainer اجرا می‌کنید

> اگر از ویندوز استفاده می‌کنید، به جای `source venv/bin/activate` دستور `venv\Scripts\activate` را وارد کنید.

> [!NOTE]
> برای پیدا کردن کلید OpenAI خود به آدرس https://platform.openai.com/settings/organization/api-keys بروید و به دنبال گزینه `API keys` بگردید. می‌توانید در همان‌جا یک کلید جدید بسازید و مقدار آن را بلافاصله کپی کنید.


۱. یک فایل *app.py* ایجاد کنید و کد زیر را در آن قرار دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

شما باید خروجی‌ای مشابه زیر را ببینید:

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```

## تمرین - ساخت یک تولیدکننده دستور پخت غذا

حالا که یک سناریو را بررسی کردیم، بیایید کدی بنویسیم که با این سناریو مطابقت داشته باشد. برای این کار، مراحل زیر را دنبال کنید:

1. از فایل *app.py* موجود به عنوان نقطه شروع استفاده کنید
1. متغیر `prompt` را پیدا کنید و کد آن را به صورت زیر تغییر دهید:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

عالی، بیایید ببینیم چطور می‌توانیم اوضاع را بهتر کنیم. برای بهتر کردن کار، می‌خواهیم مطمئن شویم کد انعطاف‌پذیر است تا مواد اولیه و تعداد دستور پخت‌ها قابل تغییر و بهبود باشند.


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

    ```python
    old_prompt_result = completion.choices[0].text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    ```
