# ساخت برنامه‌های تولید متن

تاکنون در این برنامه آموزشی دیده‌اید که مفاهیم اصلی مانند پرامپت‌ها و حتی یک رشته کامل به نام «مهندسی پرامپت» وجود دارد. بسیاری از ابزارهایی که می‌توانید با آن‌ها تعامل داشته باشید مانند ChatGPT، Office 365، Microsoft Power Platform و غیره، با استفاده از پرامپت‌ها به شما کمک می‌کنند تا کاری را انجام دهید.

برای افزودن چنین تجربه‌ای به یک برنامه، نیاز است مفاهیمی مانند پرامپت، تکمیل‌ها را درک کنید و یک کتابخانه برای کار انتخاب کنید. دقیقاً این چیزی است که در این فصل یاد خواهید گرفت.

## مقدمه

در این فصل، شما:

- با کتابخانه openai و مفاهیم اصلی آن آشنا می‌شوید.
- یک برنامه تولید متن با استفاده از openai می‌سازید.
- یاد می‌گیرید چگونه از مفاهیمی مانند پرامپت، دما و توکن‌ها برای ساختن برنامه تولید متن استفاده کنید.

## اهداف یادگیری

در پایان این درس قادر خواهید بود:

- توضیح دهید برنامه تولید متن چیست.
- یک برنامه تولید متن با استفاده از openai بسازید.
- برنامه خود را برای استفاده از تعداد بیشتر یا کمتر توکن‌ها و همچنین تغییر دما پیکربندی کنید تا خروجی متنوعی داشته باشید.

## برنامه تولید متن چیست؟

معمولاً وقتی یک برنامه می‌سازید، نوعی رابط کاربری دارد مانند موارد زیر:

- مبتنی بر فرمان. برنامه‌های کنسول برنامه‌های معمولی هستند که شما یک فرمان تایپ می‌کنید و آن کار را انجام می‌دهد. برای مثال، `git` یک برنامه مبتنی بر فرمان است.
- رابط کاربری (UI). برخی برنامه‌ها رابط‌های گرافیکی (GUI) دارند که شما دکمه‌ها را کلیک می‌کنید، متن وارد می‌کنید، گزینه‌ها را انتخاب می‌کنید و غیره.

### برنامه‌های کنسول و UI محدودیت دارند

آن را با یک برنامه مبتنی بر فرمان که در آن فرمانی را تایپ می‌کنید مقایسه کنید:

- **محدود است**. نمی‌توانید هر فرمانی را تایپ کنید، فقط فرمان‌هایی که برنامه پشتیبانی می‌کند.
- **مخصوص زبان خاص**. برخی برنامه‌ها از زبان‌های زیادی پشتیبانی می‌کنند ولی به طور پیش‌فرض برنامه برای زبان خاصی ساخته شده است، حتی اگر بتوانید پشتیبانی از زبان‌های دیگر اضافه کنید.

### مزایای برنامه‌های تولید متن

پس چه تفاوتی در برنامه تولید متن وجود دارد؟

در برنامه تولید متن، شما آزادی بیشتری دارید، محدود به مجموعه‌ای از فرمان‌ها یا زبان ورودی معینی نیستید. در عوض، می‌توانید با زبان طبیعی با برنامه تعامل داشته باشید. مزیت دیگر این است که چون شما در حال تعامل با یک منبع داده هستید که بر روی یک مجموعه وسیع اطلاعات آموزش دیده است، در حالی که یک برنامه سنتی ممکن است محدود به آنچه در یک پایگاه داده است باشد.

### با برنامه تولید متن چه چیزهایی می‌توانم بسازم؟

چیزهای زیادی می‌توان ساخت. برای مثال:

- **یک چت بات**. چت باتی که به سوالات در مورد موضوعات مختلف مانند شرکت شما و محصولات آن پاسخ می‌دهد می‌تواند گزینه خوبی باشد.
- **کمک‌کننده**. مدل‌های زبان بزرگ (LLM) در کارهایی مثل خلاصه‌سازی متن، استخراج بینش از متن، تولید متن مانند رزومه و غیره خیلی خوب هستند.
- **دستیار کدنویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید دستیار کد بسازید که در نوشتن کد کمک می‌کند. برای مثال، می‌توانید از محصولی مانند GitHub Copilot و همچنین ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چگونه شروع کنم؟

خب، باید راهی برای ادغام با یک مدل زبان بزرگ (LLM) پیدا کنید که معمولاً شامل دو رویکرد زیر است:

- استفاده از API. در اینجا درخواست‌های وب را با پرامپت خود می‌سازید و متن تولید شده را دریافت می‌کنید.
- استفاده از یک کتابخانه. کتابخانه‌ها به کپسوله کردن فراخوانی‌های API کمک می‌کنند و استفاده از آن‌ها آسان‌تر است.

## کتابخانه‌ها/SDK‌ها

چند کتابخانه شناخته‌شده برای کار با LLMها وجود دارد مانند:

- **openai**، این کتابخانه اتصال به مدل شما و ارسال پرامپت‌ها را آسان می‌کند.

سپس کتابخانه‌هایی وجود دارند که در سطح بالاتری عمل می‌کنند مانند:

- **Langchain**. Langchain شناخته‌شده است و از پایتون پشتیبانی می‌کند.
- **Semantic Kernel**. Semantic Kernel کتابخانه‌ای از مایکروسافت است که از زبان‌های C#، Python و Java پشتیبانی می‌کند.

### 2. VS Code (یا هر IDE مورد علاقه)

از پروژه Microsoft Foundry خود تمامی اطلاعات لازم برای اجرای برنامه در IDE مورد علاقه‌تان را دارید. این مثال VS Code را نشان می‌دهد.

* یک مدل مکالمه را در پروژه Foundry خود مستقر کنید. مثال‌های دما در این درس از یک مدل **Llama** مانند `Llama-3.3-70B-Instruct` استفاده می‌کنند، زیرا مدل‌های استدلالی مانند `gpt-5-mini` از `temperature`/`top_p` پشتیبانی نمی‌کنند.
* **endpoint** و **API key** پروژه را از صفحه Overview کپی کنید
* متغیرهای محیطی برای ذخیره آن‌ها ایجاد کنید: `AZURE_INFERENCE_ENDPOINT` و `AZURE_INFERENCE_CREDENTIAL` - نمونه‌هایی در bash، powershell و خط فرمان ویندوز موجود است
* وابستگی‌ها را نصب کنید: `pip install azure-ai-inference`
* کد نمونه ابتدایی را در یک فایل .py کپی کنید
* به جایی که کد شما ذخیره شده بروید و فایل را اجرا کنید: `python filename.py`

## تولید متن با ChatCompletions

روش تولید متن استفاده از کلاس `ChatCompletionsClient` است.
در `samples/python/azure_ai_inference/basic.py`، در بخش پاسخ کد، نقش کاربر را با تغییر پارامتر content به مقدار زیر به‌روزرسانی کنید:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # پارامترهای اختیاری
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

فایل به‌روزشده را اجرا کنید تا خروجی را ببینید


## انواع مختلف پرامپت‌ها، برای کارهای مختلف

اکنون دیده‌اید چطور می‌توان با استفاده از پرامپت متن تولید کرد. حتی یک برنامه در حال اجرا دارید که می‌توانید آن را تغییر دهید و انواع مختلفی از متن‌ها را ایجاد کنید.

پرامپت‌ها می‌توانند برای انواع مختلفی از کارها استفاده شوند. برای مثال:

- **تولید نوعی متن**. به عنوان مثال، می‌توانید یک شعر، سوالات برای یک آزمون و غیره تولید کنید.
- **جواب دادن به سوالات**. می‌توانید از پرامپت‌ها برای جستجوی اطلاعات مانند مثال زیر استفاده کنید: «CORS در توسعه وب به چه معناست؟».
- **تولید کد**. می‌توانید از پرامپت‌ها برای تولید کد استفاده کنید، مثلاً توسعه یک عبارت باقاعده برای اعتبارسنجی ایمیل‌ها یا چرا تمام یک برنامه، مثل یک وب اپلیکیشن را تولید نکنید؟

## تمرین: یک مولد دستور پخت

تصور کنید در خانه مواد اولیه دارید و می‌خواهید چیزی بپزید. برای این کار نیاز به یک دستور پخت دارید. راهی برای پیدا کردن دستورهای پخت، استفاده از موتور جستجو است یا می‌توانید از یک مدل زبان بزرگ (LLM) استفاده کنید.

1. از فایل موجود به عنوان نقطه شروع استفاده کنید  
1. یک متغیر `prompt` ایجاد کنید و کد نمونه را به صورت زیر تغییر دهید:  


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
