# ساخت اپلیکیشن‌های تولید متن

تا اینجا در این دوره با مفاهیم اصلی مثل پرامپت‌ها و حتی یک حوزه کامل به نام «مهندسی پرامپت» آشنا شدید. بسیاری از ابزارهایی که می‌توانید با آن‌ها کار کنید مثل ChatGPT، آفیس ۳۶۵، Microsoft Power Platform و غیره، از شما پشتیبانی می‌کنند تا با استفاده از پرامپت‌ها به هدفی برسید.

برای اینکه چنین تجربه‌ای را به اپلیکیشن خود اضافه کنید، باید مفاهیمی مثل پرامپت، completion و انتخاب یک کتابخانه مناسب را بشناسید. دقیقاً همین‌ها را در این فصل یاد می‌گیرید.

## مقدمه

در این فصل شما:

- با کتابخانه openai و مفاهیم اصلی آن آشنا می‌شوید.
- یک اپلیکیشن تولید متن با استفاده از openai می‌سازید.
- یاد می‌گیرید چطور از مفاهیمی مثل پرامپت، دما (temperature) و توکن‌ها برای ساخت اپلیکیشن تولید متن استفاده کنید.

## اهداف یادگیری

در پایان این درس، شما قادر خواهید بود:

- توضیح دهید اپلیکیشن تولید متن چیست.
- یک اپلیکیشن تولید متن با openai بسازید.
- اپلیکیشن خود را طوری تنظیم کنید که تعداد توکن‌ها را کم یا زیاد کنید و همچنین دما را تغییر دهید تا خروجی متنوعی داشته باشید.

## اپلیکیشن تولید متن چیست؟

معمولاً وقتی یک اپلیکیشن می‌سازید، یک نوع رابط کاربری دارد مثل موارد زیر:

- مبتنی بر دستور (Command-based). اپلیکیشن‌های کنسولی نمونه‌هایی هستند که شما یک دستور تایپ می‌کنید و آن کار را انجام می‌دهد. مثلاً `git` یک اپلیکیشن مبتنی بر دستور است.
- رابط کاربری (UI). بعضی اپلیکیشن‌ها رابط کاربری گرافیکی (GUI) دارند که در آن دکمه‌ها را کلیک می‌کنید، متن وارد می‌کنید، گزینه‌ها را انتخاب می‌کنید و غیره.

### اپلیکیشن‌های کنسولی و UI محدودیت دارند

مقایسه کنید با یک اپلیکیشن مبتنی بر دستور که شما یک دستور تایپ می‌کنید:

- **محدود است**. نمی‌توانید هر دستوری را تایپ کنید، فقط دستورهایی که اپلیکیشن پشتیبانی می‌کند.
- **وابسته به زبان**. بعضی اپلیکیشن‌ها از چند زبان پشتیبانی می‌کنند، اما به طور پیش‌فرض برای یک زبان خاص ساخته شده‌اند، حتی اگر بتوانید پشتیبانی زبان‌های دیگر را اضافه کنید.

### مزایای اپلیکیشن‌های تولید متن

پس اپلیکیشن تولید متن چه تفاوتی دارد؟

در یک اپلیکیشن تولید متن، انعطاف بیشتری دارید، محدود به مجموعه‌ای از دستورات یا یک زبان ورودی خاص نیستید. در عوض می‌توانید با زبان طبیعی با اپلیکیشن تعامل کنید. مزیت دیگر این است که چون با یک منبع داده تعامل دارید که روی حجم عظیمی از اطلاعات آموزش دیده، در حالی که یک اپلیکیشن سنتی ممکن است فقط به داده‌های موجود در یک پایگاه داده محدود باشد.

### با اپلیکیشن تولید متن چه چیزهایی می‌توان ساخت؟

چیزهای زیادی می‌توانید بسازید. مثلاً:

- **چت‌بات**. یک چت‌بات که به سوالات درباره موضوعاتی مثل شرکت شما و محصولاتش پاسخ می‌دهد، می‌تواند گزینه خوبی باشد.
- **دستیار**. مدل‌های زبانی بزرگ (LLM) در کارهایی مثل خلاصه‌سازی متن، استخراج بینش از متن، تولید متن مثل رزومه و غیره عالی هستند.
- **دستیار کدنویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید یک دستیار کدنویسی بسازید که به شما در نوشتن کد کمک کند. مثلاً می‌توانید از محصولاتی مثل GitHub Copilot یا ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چطور شروع کنم؟

باید راهی پیدا کنید تا با یک مدل زبانی بزرگ (LLM) ارتباط برقرار کنید که معمولاً دو روش دارد:

- استفاده از API. در این روش، درخواست‌های وب با پرامپت خود می‌سازید و متن تولیدشده را دریافت می‌کنید.
- استفاده از کتابخانه. کتابخانه‌ها فراخوانی‌های API را کپسوله می‌کنند و استفاده از آن‌ها را ساده‌تر می‌کنند.

## کتابخانه‌ها/SDKها

چند کتابخانه معروف برای کار با LLMها وجود دارد مثل:

- **openai**، این کتابخانه اتصال به مدل و ارسال پرامپت را ساده می‌کند.

همچنین کتابخانه‌هایی هستند که در سطح بالاتری کار می‌کنند مثل:

- **Langchain**. لانگ‌چین معروف است و از پایتون پشتیبانی می‌کند.
- **Semantic Kernel**. Semantic Kernel کتابخانه‌ای از مایکروسافت است که از زبان‌های C#، پایتون و جاوا پشتیبانی می‌کند.

## تولید متن با ChatCompletions

راه تولید متن استفاده از کلاس `ChatCompletionsClient` است.
در فایل `samples/python/azure_ai_inference/basic.py`، در بخش پاسخ کد، کد نقش کاربر را با تغییر پارامتر content به صورت زیر به‌روزرسانی کنید:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

فایل به‌روزشده را اجرا کنید تا خروجی را ببینید


## انواع مختلف پرامپت‌ها برای کارهای گوناگون

حالا دیدید چطور می‌شود با یک پرامپت متن تولید کرد. حتی یک برنامه دارید که می‌توانید آن را تغییر دهید تا انواع مختلفی از متن را بسازید.

پرامپت‌ها برای کارهای مختلفی کاربرد دارند. مثلا:

- **تولید یک نوع متن**. مثلا می‌توانید یک شعر بسازید، یا سوال‌هایی برای یک آزمون تولید کنید.
- **جستجوی اطلاعات**. می‌توانید با پرامپت‌ها دنبال اطلاعات بگردید، مثل این مثال: «CORS در توسعه وب یعنی چه؟»
- **تولید کد**. می‌توانید با پرامپت‌ها کد بنویسید، مثلا یک عبارت منظم برای اعتبارسنجی ایمیل بسازید یا حتی یک برنامه کامل مثل یک وب‌اپ تولید کنید.

## تمرین: تولید دستور پخت غذا

فرض کنید چند ماده غذایی در خانه دارید و می‌خواهید با آن‌ها چیزی بپزید. برای این کار به دستور پخت نیاز دارید. یکی از راه‌های پیدا کردن دستور پخت، استفاده از موتور جستجو است یا می‌توانید از یک مدل زبانی بزرگ کمک بگیرید.

می‌توانید پرامپتی مثل این بنویسید:

> «۵ دستور پخت غذایی با مواد زیر به من نشان بده: مرغ، سیب‌زمینی و هویج. برای هر دستور، همه مواد مورد استفاده را لیست کن.»

حالا که یک سناریو را بررسی کردیم، بیایید کدی بنویسیم که با سناریوی نمایش داده شده مطابقت داشته باشد. برای این کار، مراحل زیر را دنبال کنید:

1. از فایل موجود به عنوان نقطه شروع استفاده کنید
1. یک متغیر `prompt` بسازید و نمونه کد را به شکل زیر تغییر دهید:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

عالیه، بیایید ببینیم چطور می‌توانیم اوضاع را بهتر کنیم. برای بهتر شدن، می‌خواهیم مطمئن شویم که کد انعطاف‌پذیر است تا مواد اولیه و تعداد دستور پخت‌ها قابل تغییر و بهبود باشند.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
