# ساخت برنامه‌های تولید متن

تا اینجا در این برنامه درسی دیده‌اید که مفاهیم پایه‌ای مانند پرامپت‌ها وجود دارد و حتی علمی به نام «مهندسی پرامپت» وجود دارد. بسیاری از ابزارهایی که می‌توانید با آنها تعامل داشته باشید مانند ChatGPT، Office 365، Microsoft Power Platform و غیره، از پرامپت‌ها برای انجام کاری پشتیبانی می‌کنند.

برای افزودن چنین تجربه‌ای به یک برنامه، باید مفاهیمی مانند پرامپت‌ها، تکمیل‌ها را درک کرده و کتابخانه‌ای برای کار انتخاب کنید. دقیقا همین چیزی است که در این فصل خواهید آموخت.

## معرفی

در این فصل، شما:

- درباره کتابخانه openai و مفاهیم اصلی آن خواهید آموخت.
- یک برنامه تولید متن با استفاده از openai خواهید ساخت.
- خواهید فهمید چگونه از مفاهیمی مانند پرامپت، دما و توکن‌ها برای ساخت برنامه تولید متن استفاده کنید.

## اهداف یادگیری

در پایان این درس، قادر خواهید بود:

- توضیح دهید برنامه تولید متن چیست.
- یک برنامه تولید متن با استفاده از openai بسازید.
- برنامه خود را تنظیم کنید تا از توکن‌های بیشتر یا کمتر استفاده کند و همچنین دما را برای خروجی متنوع تغییر دهید.

## برنامه تولید متن چیست؟

معمولاً وقتی برنامه‌ای می‌سازید، یک نوع رابط کاربری دارد مانند موارد زیر:

- مبتنی بر دستور. برنامه‌های کنسولی برنامه‌هایی معمولی هستند که شما دستوری تایپ می‌کنید و کار را انجام می‌دهند. برای مثال، `git` یک برنامه مبتنی بر دستور است.
- رابط کاربری (UI). برخی برنامه‌ها دارای رابط‌های گرافیکی کاربری (GUI) هستند که در آن دکمه‌ها را کلیک می‌کنید، متن وارد می‌کنید، گزینه‌ها را انتخاب می‌کنید و غیره.

### برنامه‌های کنسولی و رابط کاربری محدود هستند

آن را با یک برنامه مبتنی بر دستور که در آن یک دستور تایپ می‌کنید، مقایسه کنید:

- **محدود است**. شما نمی‌توانید هر دستوری را تایپ کنید، فقط دستوراتی که برنامه پشتیبانی می‌کند.
- **مخصوص زبان است**. برخی برنامه‌ها از زبان‌های زیادی پشتیبانی می‌کنند، اما به طور پیش‌فرض برنامه برای یک زبان خاص ساخته شده است، حتی اگر بتوانید پشتیبانی زبان‌های بیشتر را اضافه کنید.

### مزایای برنامه‌های تولید متن

پس برنامه تولید متن چگونه متفاوت است؟

در برنامه تولید متن، شما انعطاف‌پذیری بیشتری دارید، محدود به مجموعه‌ای از دستورات یا زبان ورودی خاص نیستید. در عوض، می‌توانید به زبان طبیعی با برنامه تعامل داشته باشید. مزیت دیگر این است که چون شما در حال حاضر با منبع داده‌ای تعامل دارید که روی یک مجموعه بزرگ از اطلاعات آموزش دیده، در حالی که یک برنامه سنتی ممکن است محدود به محتوای موجود در یک پایگاه‌داده باشد.

### با برنامه تولید متن چه می‌توانم بسازم؟

شما می‌توانید چیزهای زیادی بسازید. برای مثال:

- **چت‌بات**. چت‌باتی که به سوالات درباره موضوعات مختلف، مانند شرکت شما و محصولاتش پاسخ می‌دهد، می‌تواند گزینه مناسبی باشد.
- **دستیار**. مدل‌های زبان بزرگ در چیزهایی مانند خلاصه‌سازی متن، استخراج بینش از متن، تولید متن مانند رزومه‌ها و غیره عالی هستند.
- **دستیار کد**. بسته به مدلی که استفاده می‌کنید، می‌توانید یک دستیار کد بسازید که به نوشتن کد کمک کند. به عنوان مثال، می‌توانید از محصولی مانند GitHub Copilot و همچنین ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چگونه می‌توانم شروع کنم؟

خب، شما باید راهی برای ادغام با یک مدل زبان بزرگ پیدا کنید که معمولاً شامل دو روش زیر است:

- استفاده از API. در اینجا شما درخواست وب را با پرامپت خود ساختار می‌دهید و متن تولید شده را دریافت می‌کنید.
- استفاده از یک کتابخانه. کتابخانه‌ها تماس‌های API را کپسوله می‌کنند و استفاده از آنها را آسان‌تر می‌کنند.

## کتابخانه‌ها/SDKها

چند کتابخانه شناخته شده برای کار با مدل‌های زبان بزرگ وجود دارد مانند:

- **openai**، این کتابخانه اتصال به مدل شما و ارسال پرامپت‌ها را آسان می‌کند.

سپس کتابخانه‌هایی هستند که روی سطح بالاتری عمل می‌کنند مانند:

- **Langchain**. Langchain شناخته شده و از پایتون پشتیبانی می‌کند.
- **Semantic Kernel**. Semantic Kernel یک کتابخانه مایکروسافت است که از زبان‌های C#، پایتون و جاوا پشتیبانی می‌کند.

## تولید متن با ChatCompletions

روش تولید متن استفاده از کلاس `ChatCompletionsClient` است.
در `samples/python/azure_ai_inference/basic.py`، در بخش پاسخ کد، پارامتر content در نقش کاربر را به شکل زیر به‌روزرسانی کنید:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # پارامترهای اختیاری
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

فایل به‌روزشده را اجرا کنید تا خروجی را مشاهده کنید


## انواع مختلف پرامپت‌ها، برای کارهای مختلف

حالا شما دیدید چطور می‌توان با استفاده از یک پرامپت متن تولید کرد. حتی شما یک برنامه دارید که اجرا شده و می‌توانید آن را تغییر دهید تا انواع مختلف متون تولید کنید.

پرامپت‌ها می‌توانند برای انواع مختلفی از وظایف استفاده شوند. برای مثال:

- **تولید نوعی متن**. برای مثال، می‌توانید شعر، سوالات یک آزمون و غیره تولید کنید.
- **جستجوی اطلاعات**. می‌توانید از پرامپت‌ها برای جستجو اطلاعات استفاده کنید، مانند مثال زیر: «CORS در توسعه وب چه معنایی دارد؟»
- **تولید کد**. می‌توانید با پرامپت کد تولید کنید، برای مثال توسعه یک عبارت منظم برای اعتبارسنجی ایمیل‌ها یا چرا کل یک برنامه مانند یک وب‌اپ را تولید نکنید؟

اکنون که یک سناریو را اجرا کرده‌ایم، بیایید کدی بنویسیم که با سناریوی نشان داده شده مطابقت داشته باشد. برای این کار مراحل زیر را دنبال کنید:

۱. از فایل موجود به عنوان نقطه شروع استفاده کنید
۲. یک متغیر `prompt` ایجاد کرده و کد نمونه را به شکل زیر تغییر دهید:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

عالی است، بیایید ببینیم چگونه می‌توانیم اوضاع را بهتر کنیم. برای بهبود اوضاع، می‌خواهیم مطمئن شویم کد انعطاف‌پذیر است، بنابراین مواد تشکیل‌دهنده و تعداد دستورهای غذایی می‌توانند بهبود یابند و تغییر کنند.


۱. بیایید کد را به شکل زیر تغییر دهیم:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
