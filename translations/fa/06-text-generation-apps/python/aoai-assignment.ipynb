## اپلیکیشن تولید متن چیست؟

معمولاً وقتی یک اپلیکیشن می‌سازید، یک نوع رابط کاربری مثل موارد زیر دارد:

- مبتنی بر فرمان. اپلیکیشن‌های کنسولی نمونه‌هایی هستند که در آن‌ها یک فرمان تایپ می‌کنید و یک کار انجام می‌شود. مثلاً `git` یک اپلیکیشن مبتنی بر فرمان است.
- رابط کاربری (UI). بعضی اپلیکیشن‌ها رابط کاربری گرافیکی (GUI) دارند که در آن‌ها دکمه‌ها را کلیک می‌کنید، متن وارد می‌کنید، گزینه‌ها را انتخاب می‌کنید و غیره.

### اپلیکیشن‌های کنسولی و UI محدودیت دارند

آن را با یک اپلیکیشن مبتنی بر فرمان مقایسه کنید که در آن یک فرمان تایپ می‌کنید:

- **محدود است**. نمی‌توانید هر فرمانی را تایپ کنید، فقط فرمان‌هایی که اپلیکیشن پشتیبانی می‌کند.
- **وابسته به زبان**. بعضی اپلیکیشن‌ها از زبان‌های مختلف پشتیبانی می‌کنند، اما به طور پیش‌فرض برای یک زبان خاص ساخته شده‌اند، حتی اگر بتوانید پشتیبانی از زبان‌های دیگر را اضافه کنید.

### مزایای اپلیکیشن‌های تولید متن

پس اپلیکیشن تولید متن چه تفاوتی دارد؟

در یک اپلیکیشن تولید متن، انعطاف بیشتری دارید، محدود به مجموعه‌ای از فرمان‌ها یا یک زبان ورودی خاص نیستید. در عوض، می‌توانید با زبان طبیعی با اپلیکیشن تعامل داشته باشید. مزیت دیگر این است که چون با یک منبع داده که روی حجم عظیمی از اطلاعات آموزش دیده کار می‌کنید، محدودیت‌های یک اپلیکیشن سنتی که فقط به دیتابیس وابسته است را ندارید.

### با اپلیکیشن تولید متن چه چیزهایی می‌توان ساخت؟

چیزهای زیادی می‌توانید بسازید. برای مثال:

- **چت‌بات**. یک چت‌بات که به سوالات درباره موضوعاتی مثل شرکت و محصولات شما پاسخ می‌دهد، می‌تواند گزینه خوبی باشد.
- **دستیار**. مدل‌های زبانی بزرگ (LLM) در کارهایی مثل خلاصه‌سازی متن، استخراج بینش از متن، تولید متن مثل رزومه و غیره عالی هستند.
- **دستیار کدنویسی**. بسته به مدل زبانی که استفاده می‌کنید، می‌توانید یک دستیار کدنویسی بسازید که به شما در نوشتن کد کمک کند. مثلاً می‌توانید از محصولاتی مثل GitHub Copilot یا ChatGPT برای کمک به نوشتن کد استفاده کنید.

## چطور می‌توانم شروع کنم؟

باید راهی برای ارتباط با یک مدل زبانی بزرگ (LLM) پیدا کنید که معمولاً شامل دو روش زیر است:

- استفاده از API. در این روش، درخواست‌های وب با پرامپت خود می‌سازید و متن تولید شده را دریافت می‌کنید.
- استفاده از کتابخانه. کتابخانه‌ها تماس‌های API را کپسوله می‌کنند و استفاده از آن‌ها را ساده‌تر می‌کنند.

### پیکربندی Azure

اگر از Azure OpenAI استفاده می‌کنید، پیکربندی به این صورت است:

```python

در بالا موارد زیر را تنظیم می‌کنیم:

- `api_key`، این همان کلید API شماست که در Azure Portal پیدا می‌کنید.

- `azure_endpoint`، این نقطه پایانی API است. می‌توانید آن را در Azure Portal کنار کلید API خود پیدا کنید.

> [!NOTE]
> `os.environ` تابعی است که متغیرهای محیطی را می‌خواند. می‌توانید از آن برای خواندن متغیرهایی مثل `AZURE_OPENAI_API_KEY` و `AZURE_OPENAI_ENDPOINT` استفاده کنید.

## تولید متن

روش تولید متن استفاده از کلاس `chat.completion` است. مثال زیر را ببینید:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

در کد بالا، یک شی completion می‌سازیم و مدل مورد نظر و پرامپت را به آن می‌دهیم. سپس متن تولید شده را چاپ می‌کنیم.

### تکمیل چت (Chat completions)

تا اینجا دیدید که چطور از `Completion` برای تولید متن استفاده کردیم. اما یک کلاس دیگر به نام `ChatCompletion` وجود دارد که برای چت‌بات‌ها مناسب‌تر است. مثال زیر را ببینید:

```python
client = AzureOpenAI(
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT'), 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-05-15"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

در فصل بعدی بیشتر درباره این قابلیت صحبت خواهیم کرد.

## تمرین - اولین اپلیکیشن تولید متن شما

حالا که یاد گرفتیم چطور سرویس Azure OpenAI را راه‌اندازی و پیکربندی کنیم، وقت آن است که اولین اپلیکیشن تولید متن خود را بسازید. برای ساخت اپ خود، این مراحل را دنبال کنید:


۱. یک محیط مجازی بسازید و openai را نصب کنید:

  > [!NOTE] این مرحله لازم نیست اگر این نوت‌بوک را در Codespaces یا داخل یک Devcontainer اجرا می‌کنید


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> اگر از ویندوز استفاده می‌کنید، به جای `source venv/bin/activate` دستور `venv\Scripts\activate` را وارد کنید.

> [!NOTE]
> برای پیدا کردن کلید Azure OpenAI خود، به https://portal.azure.com/ بروید، عبارت `Open AI` را جستجو کنید و سپس `Open AI resource` را انتخاب کنید. بعد از آن روی `Keys and Endpoint` کلیک کنید و مقدار `Key 1` را کپی کنید.


۱. یک فایل *app.py* ایجاد کنید و کد زیر را در آن قرار دهید:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

شما باید خروجی‌ای مشابه زیر را ببینید:

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```

- **تولید کد**. می‌توانید با پرامپت‌ها کد بنویسید، مثلا یک عبارت منظم برای اعتبارسنجی ایمیل بسازید یا حتی یک برنامه کامل مثل یک اپلیکیشن وب تولید کنید.

1. متغیر `prompt` را پیدا کنید و کد آن را به صورت زیر تغییر دهید:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

عالی است، بیایید ببینیم چطور می‌توانیم اوضاع را بهتر کنیم. برای بهبود کار، می‌خواهیم مطمئن شویم که کد انعطاف‌پذیر است تا مواد اولیه و تعداد دستور پخت‌ها بتوانند به راحتی تغییر کنند و بهبود یابند.


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

    همانطور که می‌بینید، هر دستوری که شیر داشته باشد حذف شده است. اما اگر به لاکتوز حساسیت دارید، شاید بخواهید دستورهایی که پنیر دارند را هم حذف کنید، پس باید دقیق باشید.

    ```python
    
- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = completion.choices[0].text
    prompt = "یک لیست خرید برای دستورهای تولید شده بساز و لطفاً موادی که قبلاً دارم را وارد نکن."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
    
    # چاپ پاسخ
    print("لیست خرید:")
    print(completion.choices[0].message.content)

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python

        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_tokens` is 1200.

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be.    To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:
