## Генериране на текст

Начинът да генерирате текст е да използвате метода `responses.create`. Ето един пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В горния код създаваме обект отговор и подаваме модела, който искаме да използваме, както и подсказката. След това отпечатваме генерирания текст.

### Отговори в чат

До момента видяхте как можем да генерираме текст от една подсказка. Responses API е също подходящ за чатботове - подавате списък от съобщения като `input`, за да създадете разговор. Ето пример за използване:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Повече за тази функционалност ще научите в следваща глава.

## Упражнение - създайте генератор на рецепти

Сега, след като изиграхме един сценарий, нека напишем код, който отговаря на демонстрирания сценарий. За да го направите, следвайте следните стъпки:

1. Използвайте съществуващия файл *app.py* като отправна точка
1. Намерете променливата `prompt` и променете нейния код на следното:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> ЗАБЕЛЕЖКА, вашият LLM е недетерминистичен, така че може да получите различни резултати всеки път, когато стартирате програмата.

Отлично, нека видим как можем да подобрим нещата. За да подобрим нещата, искаме да се уверим, че кодът е гъвкав, така че съставките и броят на рецептите могат да бъдат подобрявани и променяни.


1. Нека променим кода по следния начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
