## Генериране на текст

Начинът за генериране на текст е чрез метода `responses.create`. Ето пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В горния код създаваме обект отговор и задаваме модела, който искаме да използваме, както и подсказката. След това отпечатваме генерирания текст.

### Отговори в чат

Досега видяхте как можем да генерираме текст от една подсказка. API-то Responses е също подходящо за чатботове – предоставяте списък със съобщения като `input`, за да изградите разговор. Ето пример за използването му:

## Упражнение - създаване на генератор на рецепти

Сега, след като изиграхме сценария, нека напишем код, който съответства на демонстрирания сценарий. За да го направите, следвайте тези стъпки:

1. Намерете променливата `prompt` и променете нейния код на следното:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> ЗАБЕЛЕЖКА, вашият LLM е недетерминиран, така че може да получите различни резултати всеки път, когато стартирате програмата.

Страхотно, нека видим как можем да подобрим нещата. За да подобрим нещата, искаме да сме сигурни, че кодът е гъвкав, така че съставките и броят на рецептите да могат да се подобряват и променят.


1. Нека променим кода по следния начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
