### Конфигуриране на Azure

Ако използвате Azure OpenAI, ето как настройвате конфигурацията. Responses API се обслужва от Azure OpenAI **v1 крайна точка**, така че насочваме клиента `OpenAI` към `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

По-горе настройваме следното:

- `api_key`, това е вашият API ключ, открит в Azure Portal.
- `base_url`, това е вашата крайна точка Azure OpenAI с добавен `/openai/v1/`. Можете да намерите крайна точка в Azure Portal до вашия API ключ. Използването на v1 крайна точка означава, че вече не е необходимо да подавате `api_version`.
- `deployment`, това е името на разгръщането на модела, което сте създали в портала Foundry.

> [!NOTE]
> `os.environ` е карта, която чете променливи на средата. Можете да я използвате за четене на променливи на средата като `AZURE_OPENAI_API_KEY` и `AZURE_OPENAI_ENDPOINT`.

## Генериране на текст

Начинът за генериране на текст е чрез използване на метода `responses.create`. Ето един пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В горния код създаваме обект на отговор и подаваме модела, който искаме да използваме и подканата. След това отпечатваме генерирания текст.

Сега, когато вече изиграхме един сценарий, нека напишем код, който съответства на показания сценарий. За да го направите, следвайте следните стъпки:

1. Намерете променливата `prompt` и сменете нейния код със следното:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Отлично, нека видим как можем да подобрим нещата. За да подобрим нещата, искаме да сме сигурни, че кодът е гъвкав, така че съставките и броят на рецептите да могат да бъдат подобрявани и променяни.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
