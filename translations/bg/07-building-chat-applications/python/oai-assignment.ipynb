### 2. Импортирайте помощни библиотеки и създайте идентификационни данни


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Намиране на правилния модел

```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

#### Как да контролираме голям езиков модел  
"Сред всички входове към голям езиков модел, с разлика най-влиятелен е текстовият подсещател(1).

Големите езикови модели могат да бъдат подсещани да произвеждат изход по няколко начина:

Инструкция: Кажете на модела какво искате
Допълнение: Предизвикайте модела да допълни началото на това, което искате
Демонстрация: Покажете на модела какво искате, с помощта на:

#### Има три основни насоки за създаване на подсещания:

**Покажи и кажи**. Направете ясно какво искате чрез инструкции, примери или комбинация от двете. Ако искате моделът да подреди списък с елементи в азбучен ред или да класифицира параграф по настроение, покажете му, че това е вашето желание.

**Осигурете качествени данни**. Ако се опитвате да създадете класификатор или да накарате модела да следва определен модел, уверете се, че има достатъчно примери. Винаги проверявайте примерите си — моделът обикновено е достатъчно умен да пропусне основни правописни грешки и да ви даде отговор, но може да приеме това за намерено и да повлияе на отговора.

**Проверете настройките си.** Параметрите temperature и top_p контролират доколко детерминистичен е моделът при генериране на отговор. Ако задавате въпрос, където има само един правилен отговор, желаете тези параметри да са по-ниски. Ако искате по-разнообразни отговори, може да ги зададете по-високи. Най-честата грешка е да се смята, че тези настройки контролират „умното“ или „креативното“.

### 5. Изпратете!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

Тук показваме, че увеличаването на езиковите модели значително подобрява изпълнението по принципно независими от задачата малки серии от примери, понякога дори достигащо конкурентоспособност с предишни съвременни подходи за финно настройване.
