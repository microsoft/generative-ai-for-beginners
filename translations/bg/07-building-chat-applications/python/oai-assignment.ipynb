### 2. Импортиране на помощни библиотеки и създаване на идентификационни данни


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Намиране на правилния модел  
Модели като GPT-4o и GPT-4o mini могат да разбират и генерират естествен език и са налични на платформата OpenAI с различни нива на мощност и скорост, подходящи за различни задачи.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Проектиране на подсказки  

"Магията на големите езикови модели е, че чрез обучение за минимизиране на тази грешка при предсказване върху огромно количество текст, моделите в крайна сметка научават концепции, полезни за тези предсказвания. Например, те научават концепции като"(1):

* как да се пише код
* и т.н.

#### Как да контролираме голям езиков модел  
"От всички входни данни към голям езиков модел, най-влиятелният е без съмнение текстовата подсказка(1).

Големите езикови модели могат да бъдат подканяни да генерират изход по няколко начина:

Инструкция: Кажете на модела какво искате
Допълване: Подтикнете модела да завърши началото на това, което искате
Демонстрация: Покажете на модела какво искате, с:
Няколко примера в подсказката
Много стотици или хиляди примери в датасет за донастройка"



#### Има три основни насоки при създаване на подсказки:

**Покажи и кажи**. Направете ясно какво искате чрез инструкции, примери или комбинация от двете. Ако искате моделът да подреди списък от елементи в азбучен ред или да класифицира параграф по настроение, покажете му, че това искате.

**Осигурете качествени данни**. Ако се опитвате да изградите класификатор или искате моделът да следва определен образец, уверете се, че има достатъчно примери. Преглеждайте внимателно примерите си — моделът обикновено е достатъчно умен, за да прозре основни правописни грешки и да ви даде отговор, но също така може да предположи, че това е умишлено и това да повлияе на отговора.

**Проверете настройките си.** Параметрите temperature и top_p контролират колко детерминистичен е моделът при генериране на отговор. Ако искате отговор, за който съществува само един правилен отговор, трябва да ги настроите по-ниско. Ако искате по-разнообразни отговори, може да ги зададете по-високо. Най-честата грешка, която хората правят с тези настройки, е да предполагат, че те контролират „умност“ или „креативност“.


Източник: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Изпращане!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
