### 2. Impordi abiteegid ja loo volitused


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Õige mudeli leidmine  
Mudelid nagu GPT-4o ja GPT-4o mini suudavad mõista ja genereerida loomulikku keelt ning on saadaval OpenAI platvormil erineva võimsuse ja kiirusega, mis sobivad erinevateks ülesanneteks.


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

## 4. Käsu kujundus  

"Suurte keelemudelite võlu seisneb selles, et treenides neid selle ennustusvea minimeerimiseks tohutul hulgal tekstil, õpivad mudelid lõpuks mõisteid, mis on nendes ennustustes kasulikud. Näiteks õpivad nad mõisteid nagu"(1):

#### Kuidas suunata suurt keelemudelit  
"Kõigist suurte keelemudelite sisenditest on mõjuvõimsaim tekstikäsu prompt(1).

Suurte keelemudelitele saab väljundi genereerimiseks anda käske mõnel viisil:

Juhend: Ütle mudelile, mida sa tahad
Täiendamine: Julgusta mudelit lõpetama seda, mida sa tahad alustada
Näide: Näita mudelile, mida sa tahad, kas:
Mõned näited käsus
Sadade või tuhandete näidete kasutamine peenetreenimise andmestikus"



#### On kolm põhilist juhist promptide loomiseks:

**Näita ja ütle**. Tee selgeks, mida sa tahad, kas juhiste, näidete või nende kahe kombinatsiooni kaudu. Kui sa tahad, et mudel järjestaks loendi elemente tähestikulises järjekorras või klassifitseeriks lõiku meeleolu järgi, näita talle, et see on see, mida sa tahad.

**Pakku kvaliteetset andmestikku**. Kui sa üritad ehitada klassifikaatorit või saada mudelit järgima mustrit, veendu, et näiteid on piisavalt. Kindlasti korrektuuri oma näiteid — mudel on tavaliselt piisavalt tark, et näha läbi lihtsate õigekirjavigade ja anda sulle vastus, kuid ta võib ka eeldada, et see on tahtlik ja see võib vastusele mõju avaldada.

**Kontrolli oma seadeid.** Temperatuuri ja top_p seaded kontrollivad, kui deterministlik mudel vastuse genereerimisel on. Kui sa küsid vastust, kus on ainult üks õige, siis peaksid need sätted olema madalamad. Kui sa soovid mitmekesisemaid vastuseid, võid neid määrata kõrgemaks. Kõige tavalisem viga nende seadete puhul on arvata, et need kontrollivad "nutikust" või "loovust".


Allikas: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Esita! 


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
