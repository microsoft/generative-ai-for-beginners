### 2. Impordi abiteegid ja loo mandaadid


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Õige mudeli leidmine  
Mudelid nagu GPT-4o ja GPT-4o mini suudavad mõista ja genereerida loomulikku keelt ning on saadaval OpenAI platvormil erineva võimsuse ja kiirusega, mis sobivad erinevateks ülesanneteks.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Päringu kujundamine  

"Suure keelemudelite võlu seisneb selles, et treenides neid selle ennustuse vea minimeerimiseks tohututes tekstimahtudes, õpivad mudelid lõpuks ära kontseptsioonid, mis on nende ennustuste tegemiseks kasulikud. Näiteks õpivad nad kontseptsioone nagu"(1):

#### Kuidas juhtida suurt keelemudelit  
"Kõikidest suurt keelemudelisse sisenditest on ülimalt mõjukas tekstipäring(1).

Suuri keelemudeleid saab pärida paaril viisil, et nad annaksid vastuseid:

Juhis: Ütle mudelile, mida sa tahad
Täiendamine: Püüa suunata mudelit lõpetama seda, mida sa tahad alustada
Demonstratsioon: Näita mudelile, mida sa tahad, kas:
Mõned näited päringus
Korraga sadu või tuhandeid näiteid peenhäälestusandmestikus"



#### Päringute loomisel on kolm põhireeglit:

**Näita ja ütle**. Tee selgeks, mida sa tahad, kas juhiste, näidete või nende kahe kombinatsiooniga. Kui tahad, et mudel järjestaks esemete nimekirja tähestikulises järjekorras või klassifitseeriks lõiku meeleolu järgi, näita talle, et see on sinu soov.

**Pakkuge kvaliteetset andmestikku**. Kui püüad luua klassifikaatorit või suunata mudelit mustrit järgima, veendu, et on piisavalt näiteid. Kontrolli kindlasti oma näiteid — mudel on tavaliselt piisavalt tark, et märgata lihtsaid õigekirjavigu ja anda vastus, aga võib ka eeldada, et see on tahtlik ja see võib vastust mõjutada.

**Kontrolli seadeid.** Temperatuuri ja top_p seaded kontrollivad, kui deterministlik mudel on vastuse genereerimisel. Kui soovid vastust, millel on ainult üks õige lahendus, siis soovid neid seadeid madalamale panna. Kui soovid mitmekesisemaid vastuseid, võid seadeid kõrgemale seada. Kõige levinum viga nende seadete puhul on arvata, et need on "nutikuse" või "loomingulisuse" reguleerijad.


Allikas: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Esita!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

Seevastu inimesed saavad üldjuhul sooritada uut keeleülesannet vaid mõne näite või lihtsate juhiste põhjal – midagi, mis on tänapäeva NLP süsteemidele endiselt suureks väljakutseks. Siin näitame, et keelemudelite skaleerimine parandab märkimisväärselt ülesande suhtes neutraalset few-shot sooritust, mõnikord jõudes isegi varasemate tipptasemel täiendõppe meetodite tasemele.
