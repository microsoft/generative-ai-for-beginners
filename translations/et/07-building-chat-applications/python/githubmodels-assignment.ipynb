### 2. Impordi abiteegid ja loo mandaadid


```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)
```

### 3. Õige mudeli leidmine  
Mudelid nagu GPT-4o ja GPT-4o mini mõistavad ja genereerivad loomulikku keelt ning on saadaval Microsoft Foundry Models kataloogis koos Meta, Mistrali, Cohere ja Microsofti mudelitega.


```python
# Select a general purpose chat model
model_name = "gpt-5-mini"
```

## 4. Käskluse kujundamine  

"Suurte keelemudelite võlu seisneb selles, et õppides vähendama seda prognoosimisviga tohututes tekstikogustes, omandavad mudelid teadmised, mis on kasulikud nende prognooside tegemiseks. Näiteks omandavad nad mõisted nagu"(1):

#### Kuidas suunata suurt keelemudelit  
"Kõikide sisendite seas suurte keelemudelitele on kõige mõjukam tekstikäsklus(1).

Suurte keelemudelitega saab töödelda väljundit mitmel viisil:

Juhis: Ütle mudelile, mida sa tahad
Lõpetus: Pane mudel lõpetama seda, mida sa soovisid alustada
Demostratsioon: Näita mudelile, mida sa tahad, kas:
Mõned näited käskluses
Sadade või tuhandete näidete kasutamine peenhäälestamiseks treeningandmetes"



#### Kolm põhijuhist käskluste loomiseks:

**Näita ja ütle.** Tee selgeks, mida sa tahad, kasutades juhiseid, näiteid või mõlema kombinatsiooni. Kui tahad, et mudel järjestaks esemeid tähestikulises järjekorras või klassifitseeriks lõiku meeleolu järgi, näita talle, et see on su soov.

**Esita kvaliteetset andmestikku.** Kui proovid luua klassifikaatorit või panna mudeli jälgima mustrit, veendu, et on piisavalt näiteid. Kontrolli oma näiteid hoolikalt — mudel on tavaliselt piisavalt tark, et näha lihtsaid õigekirjavigu ja anda vastus, kuid võib ka eeldada, et need on tahtlikud, mis võib vastust mõjutada.

**Kontrolli oma seadeid.** Temperatuuri ja top_p seaded määravad, kui deterministlik mudel on vastuse genereerimisel. Kui küsid vastust, millel on ainult üks õige lahendus, tuleks need sätted madalamale seada. Kui otsid mitmekesisemaid vastuseid, võid need seada kõrgemale. Kõige saglem viga nende sätete kasutamisel on arvamine, et need kontrollivad mudeli "nutikust" või "loovust".


Allikas: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Esita! 


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.complete(
  model=model_name,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```
