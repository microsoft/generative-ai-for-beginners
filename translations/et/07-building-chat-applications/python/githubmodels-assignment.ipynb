### 2. Impordi abiraamatukogud ja loo volitused


```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)
```

### 3. Õige mudeli leidmine  
Mudelid nagu GPT-4o ja GPT-4o mini suudavad mõista ja genereerida loomulikku keelt ning on saadaval Microsoft Foundry mudelite kataloogis koos Metalt, Mistralilt, Cohere'ilt ja Microsoftilt pärit mudelitega.


```python
# Select a general purpose chat model
model_name = "gpt-4o-mini"
```

## 4. Päringukujundus  

"Suurte keelemudelite võlu peitub selles, et neid treenitakse vähendama seda ennustusviga tohututes tekstimahtudes, mille tulemusena õpivad mudelid mõisteid, mis on nende ennustuste jaoks kasulikud. Näiteks õpivad nad mõisteid nagu"(1):

#### Kuidas juhtida suurt keelemudelit  
"Kõikidest suurte keelemudelite sisenditest on kõige mõjukam tekstipäring(1).

Suuri keelemudeleid saab panna väljundit tootma mitmel viisil:

Juhend: Ütle mudelile, mida soovid
Lõpetamine: Pane mudel lõpetama selle, mille alguse sa tahad
Näitamine: Näita mudelile, mida soovid, kas siis:
Mõned näited päringus
Sadu või tuhandeid näiteid peenhäälestamise andmekogus"



#### Kolm põhimõttelist juhist päringute koostamiseks:

**Näita ja ütle**. Tee selgeks, mida sa tahad, olgu see siis juhtnööride, näidete või mõlema kombinatsiooni kaudu. Kui tahad mudelit järjestama alapealkirju tähestikulises järjekorras või klassifitseerima lõiku meeleolu järgi, näita talle, et see on see, mida sa tahad.

**Pakku kvaliteetset andmestikku**. Kui üritad luua klassifikaatorit või paned mudelit järgima kindlat mustrit, veendu, et näiteid oleks piisavalt. Kontrolli kindlasti oma näiteid üle — mudel on tavaliselt piisavalt nutikas, et läbi näha põhilised õigekirjavead ja anda vastus, kuid ta võib ka eeldada, et see on tahtlik ja see võib vastust mõjutada.

**Kontrolli oma seadeid.** Temperatuuri ja top_p seaded kontrollivad, kui deterministlik mudel vastust genereerib. Kui küsid vastust, millel on ainult üks õige vastus, siis tahad neid seadeid madalamale panna. Kui soovid mitmekesisemaid vastuseid, võid need seaded kõrgemale panna. Kõige tavalisem viga, mida inimesed nende seadistustega teevad, on eeldada, et need on "nutikuse" või "loovuse" kontrollid.


Allikas: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Esita!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.complete(
  model=model_name,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```
