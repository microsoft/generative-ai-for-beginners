## Mis on tekstigeneratsiooni rakendus?

Tavaliselt, kui ehitad rakendust, on sellel mingi kasutajaliides, näiteks järgmine:

- Käsurea põhine. Konsoolirakendused on tüüpilised rakendused, kus sa sisestad käsu ja see täidab ülesande. Näiteks on `git` käsurea põhine rakendus.
- Kasutajaliides (UI). Mõnel rakendusel on graafilised kasutajaliidesed (GUI), kus sa klikid nuppe, sisestad teksti, valid valikuid ja palju muud.

### Konsooli- ja GUI-rakendused on piiratud

Võrdle seda käsurea rakendusega, kus sa kirjutad käsu:

- **See on piiratud**. Sa ei saa lihtsalt kirjutada ükskõik millist käsku, vaid ainult neid, mida rakendus toetab.
- **Keelepõhine**. Mõned rakendused toetavad mitut keelt, kuid vaikimisi on rakendus ehitatud kindlale keelele, isegi kui saad lisada rohkem keeletoetust.

### Tekstigeneratsiooni rakenduste eelised

Kuidas siis tekstigeneratsiooni rakendus erineb?

Tekstigeneratsiooni rakenduses on sul rohkem paindlikkust, sa ei ole piiratud kindlate käskude või konkreetse sisendkeelega. Selle asemel saad kasutada loomulikku keelt rakendusega suhtlemiseks. Teine eelis on see, et sa juba suhtled andmeallikaga, mis on koolitatud väga suure infohulga peal, samas kui traditsiooniline rakendus võib olla piiratud ainult andmebaasis oleva infoga.

### Mida ma saan tekstigeneratsiooni rakendusega ehitada?

On palju asju, mida saad ehitada. Näiteks:

- **Vestlusrobot**. Vestlusrobot, mis vastab küsimustele teemadel nagu sinu ettevõte ja selle tooted, võib olla hästi sobiv.
- **Abiagent**. LLM-id on head asju nagu teksti kokkutõmbamine, tekstist teadmiste hankimine, tekstide loomine nagu CV-d ja palju muud.
- **Koodi assistent**. Sõltuvalt kasutatavast keelemudelist saad ehitada koodi assistendi, mis aitab sul koodi kirjutada. Näiteks võid kasutada tooteid nagu GitHub Copilot või ChatGPT, et aidata koodi kirjutada.

## Teksti genereerimine

Teksti genereerimiseks kasutame meetodit `responses.create`. Näide:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Ülalolevas koodis loome vastuse objekti, edastame mudeli, mida soovime kasutada ja prompti. Seejärel prindime genereeritud teksti.

### Vestluse vastused

Seni oled näinud, kuidas saab genereerida teksti ühe promptiga. Responses API sobib ka vestlusrobotite jaoks – sa annad sisendiks vestlussoovide nimekirja, et ehitada üles vestlust. Näide selle kasutamisest:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Selle funktsionaalsuse kohta rohkem järgmistes peatükkides.

- **Sõna tokenite pikkuse kohta**. Peaksime kaaluma, kui palju tokeneid vajame soovitud teksti genereerimiseks. Tokenid maksavad raha, nii et seal, kus võimalik, peaksime püüdma olla säästlik kasutatavate tokenite arvu osas. Näiteks, kas me saame prompti sõnastada nii, et kasutaksime vähem tokeneid?

   Tokenite arvu muutmiseks saab kasutada parameetrit `max_output_tokens`. Näiteks, kui soovite kasutada 100 tokenit, teeksite järgmist:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Temperatuuriga katsetamine**. Temperatuur on midagi, millest pole seni räägitud, kuid see on oluline kontekst programmi käitumise jaoks. Mida kõrgem temperatuur, seda rohkem juhuslik on väljund. Vastupidi, madalam temperatuur teeb väljundi ettearvatavamaks. Mõtle, kas soovid oma väljundis varieeruvust või mitte.

   Temperatuuri muutmiseks saab kasutada parameetrit `temperature`. Näiteks, kui soovid kasutada temperatuuri 0.5, teeksid järgmist:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Märkus, mida lähemal 1.0, seda varieeruvam väljund.



## Kodune ülesanne

Selle ülesande raames saad valida, mida ehitada.

- Ajaloo bot – tee ajalugu elavaks, pane bot mängima kindlat ajaloolist tegelast ja esita küsimusi tema elu ja ajastu kohta.
