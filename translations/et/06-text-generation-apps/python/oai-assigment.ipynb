## Mis on teksti genereerimise rakendus?

Tavaliselt, kui sa ehitad rakendust, on sellel mingi kasutajaliides nagu alljärgnev:

- Käskluspõhine. Konsoolirakendused on tüüpilised rakendused, kus trükkid käsu ja see täidab ülesande. Näiteks `git` on käskluspõhine rakendus.
- Kasutajaliides (UI). Mõnel rakendusel on graafilised kasutajaliidesed (GUI), kus klikid nuppudele, sisestad teksti, valid valikuid ja palju muud.

### Konsooli- ja kasutajaliidese rakendused on piiratud

Võrdle seda käskluspõhise rakendusega, kus sa trükid käsu:

- **See on piiratud**. Sa ei saa lihtsalt sisestada ükskõik millist käsku, vaid ainult neid, mida rakendus toetab.
- **Keelepõhine**. Mõned rakendused toetavad mitmeid keeli, kuid vaikimisi on rakendus loodud konkreetse keele jaoks, isegi kui saad lisada rohkem keelte tuge.

### Teksti genereerimise rakenduste eelised

Kuidas siis teksti genereerimise rakendus erineb?

Teksti genereerimise rakenduses on sul rohkem paindlikkust, sa ei ole piiratud kindlate käskude või konkreetse sisendkeelega. Selle asemel saad kasutada loomulikku keelt rakendusega suhtlemiseks. Teine eelis on see, et sa juba suhtled andmeallikaga, mis on koolitatud tohutul hulgal informatsioonil, samal ajal kui traditsiooniline rakendus võib olla piiratud sellega, mis on andmebaasis.

### Mida ma saan teksti genereerimise rakendusega luua?

Sa saad luua palju asju. Näiteks:

- **Vestlusrobot**. Vestlusrobot, mis vastab küsimustele erinevatel teemadel, näiteks sinu ettevõtte ja selle toodete kohta, võiks olla väga sobiv.
- **Abiline**. LLM-id on suurepärased asjades nagu teksti kokkuvõtete tegemine, tekstist teadmiste saamine, tekstide loomine nagu CV-d ja palju muud.
- **Koodi assistent**. Sõltuvalt kasutatavast keelemudelist saad teha koodi assistendi, kes aitab sul koodi kirjutada. Näiteks võid kasutada tooteid nagu GitHub Copilot või ChatGPT, mis aitavad sul koodi kirjutada.

## Teksti genereerimine

Teksti genereerimiseks kasutatakse `responses.create` meetodit. Näide:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Ülaltoodud koodis loome vastuse objekti, määrame mudeli ja prompti ning prindime genereeritud teksti.

### Vestluse vastused

Seni nägid, kuidas genereerida teksti ühe promptiga. Responses API sobib hästi ka vestlusrobotite jaoks — annad `input`-iks sõnumite loendi, et üles ehitada vestlus. Näide selle kasutamisest:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Selle funktsionaalsuse kohta rohkem tulevas t peatükis.

- **Sõna tokenite pikkusest**. Peaksime kaaluma, kui palju tokeneid meil on vaja soovitud teksti genereerimiseks. Tokenid maksavad raha, seega peaksime võimalikult väikese tokenite arvuga olema kokkuhoidlikud. Näiteks, kas saame prompti sõnastada nii, et kasutada vähem tokeneid?

   Tokenite arvu muutmiseks saab kasutada `max_output_tokens` parameetrit. Näiteks, kui tahad kasutada 100 tokenit, teeksid nii:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Temperatuuriga eksperimenteerimine**. Temperatuur on midagi, mida me pole seni maininud, kuid see on oluline kontekst meie programmi käitumiseks. Mida kõrgem temperatuuriväärtus, seda juhuslikum on väljund. Mida madalam temperatuuriväärtus, seda ettearvatavam on väljund. Mõtle, kas soovid oma väljundis varieeruvust või mitte.

   Temperatuuri muutmiseks võid kasutada `temperature` parameetrit. Näiteks, kui soovid kasutada temperatuuri 0.5, teeksid nii:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Märkus, mida lähemal 1.0-le, seda varieeruvam on väljund.
