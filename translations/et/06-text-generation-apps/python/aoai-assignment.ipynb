## Mis on tekstigeneratsiooni rakendus?

Tavaliselt, kui sa ehitad rakenduse, on sellel mingi kasutajaliides nagu alljärgnev:

- Käsklustel põhinev. Konsoolirakendused on tüüpilised rakendused, kus sa kirjutad käsu ja see täidab ülesande. Näiteks `git` on käsklustel põhinev rakendus.
- Kasutajaliides (UI). Mõnel rakendusel on graafiline kasutajaliides (GUI), kus sa klikid nuppe, sisestad teksti, valid valikuid ja muud sellist.

### Konsooli- ja UI-rakendused on piiratud

Võrdle seda käsklustel põhineva rakendusega, kuhu kirjutad käsu:

- **See on piiratud**. Sa ei saa kirjutada suvalist käsku, vaid ainult neid, mida rakendus toetab.
- **Keele-spetsiifiline**. Mõned rakendused toetavad mitut keelt, kuid vaikimisi on rakendus ehitatud konkreetsele keelele, kuigi saab lisada ka teisi keeli.

### Tekstigeneratsiooni rakenduste eelised

Kuidas on tekstigeneratsiooni rakendus siis erinev?

Tekstigeneratsiooni rakenduses on sul rohkem paindlikkust, sa ei ole piiratud kindlate käskude ega konkreetse sisendkeelega. Selle asemel võid kasutada looduskeelt rakendusega suhtlemiseks. Teine eelis on see, et juba suhtled andmeallikaga, mis on treenitud tohutul hulgal teabel, samas kui traditsiooniline rakendus võib olla piiratud selle kaudu, mis on andmebaasis.

### Mida ma saan tekstigeneratsiooni rakendusega ehitada?

On palju asju, mida saad ehitada. Näiteks:

- **Chatbot**. Chatbot, mis vastab küsimustele teemade kohta, nagu sinu ettevõte ja selle tooted, võib olla sobiv valik.
- **Abi**. LLMid on suurepärased tekstide kokkuvõtete tegemisel, teksti analüüsis, tekstide loomisel nagu CV-d ja muus osas.
- **Koodi assistent**. Sõltuvalt sellest, millist keelemudelit kasutad, saad ehitada koodiassistendi, mis aitab sul koodi kirjutada. Näiteks võid kasutada tooteid nagu GitHub Copilot või ChatGPT, et koodi kirjutamisel abiks olla.

### Sea konfigureerimine Azure jaoks

Kui kasutad Azure OpenAI-d, siis seadista konfigureerimine nii. Responses API teenindatakse Azure OpenAI **v1 lõpp-punktist**, seega suuname `OpenAI` kliendi aadressile `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

```

Ülal seadsime järgmist:

- `api_key`, see on sinu API võti, mis leiab Azure portaalist.
- `base_url`, see on sinu Azure OpenAI lõpp-punkt, millele on lisatud `/openai/v1/`. Lõpp-punkti leidmiseks vaata Azure portaali oma API võtmest kõrval. V1 lõpp-punkti kasutamine tähendab, et `api_version` parameetrit enam saata ei pea.
- `deployment`, see on mudeli paigutuse nimi, mille lõid Foundry portaalis.

> [!NOTE]
> `os.environ` on kaardistus, mis loeb keskkonnamuutujaid. Seda saad kasutada, et lugeda keskkonnamuutujaid nagu `AZURE_OPENAI_API_KEY` ja `AZURE_OPENAI_ENDPOINT`.

## Genereri tekst

Teksti genereerimiseks kasuta meetodit `responses.create`. Näide:

```python

```

### Improve by adding filter and shopping list

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
