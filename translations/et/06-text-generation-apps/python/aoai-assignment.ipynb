## Mis on tekstigeneratsiooni rakendus?

Tavaliselt, kui sa ehitad rakenduse, on sellel mingi kasutajaliides, mis võib olla näiteks:

- Käskluspõhine. Konsoolirakendused on tüüpilised rakendused, kus sa tippid käsu ja see täidab mingi ülesande. Näiteks on `git` käskluspõhine rakendus.
- Kasutajaliides (UI). Mõnel rakendusel on graafiline kasutajaliides (GUI), kus sa klikid nuppudele, sisestad teksti, valid valikuid ja muud sellist.

### Konsooli- ja UI-rakendused on piiratud

Võrdle seda käskluspõhise rakendusega, kus sa tippid käsu:

- **See on piiratud**. Sa ei saa lihtsalt tippida suvalist käsku, vaid ainult neid, mida rakendus toetab.
- **Keelepõhine**. Mõned rakendused toetavad paljusid keeli, kuid vaikimisi on rakendus ehitatud kindla keele jaoks, isegi kui saad lisada rohkem keeletoetust.

### Tekstigeneratsiooni rakenduste eelised

Niisiis, mis teeb tekstigeneratsiooni rakenduse teistsuguseks?

Tekstigeneratsiooni rakenduses on sul suurem paindlikkus, sa ei ole piiratud kindla käskude komplektiga ega kindla sisendkeelega. Selle asemel saad kasutada loomulikku keelt, et suhelda rakendusega. Veelgi enam, kuna sa juba suhtled andmeallikaga, mis on koolitatud tohutu infohulga peal, on see oluliselt rikkalikum kui traditsiooniline rakendus, mis on piiratud andmebaasis oleva sisuga.

### Mida ma saan tekstigeneratsiooni rakendusega ehitada?

On palju asju, mida saad teha. Näiteks:

- **Vestlusrobot**. Vestlusrobot, mis vastab küsimustele teemadel nagu sinu ettevõte ja selle tooted, võiks olla hea valik.
- **Abiline**. Suured keelemudelid (LLMid) sobivad suurepäraselt tekstide kokkuvõtete tegemiseks, tekstist teadmiste saamiseks, tekstide koostamiseks nagu CV-d jms.
- **Koodiabiline**. Sõltuvalt kasutatavast keelemudelist saad luua koodiabilise, mis aitab sul koodi kirjutada. Näiteks võid kasutada selliseid tooteid nagu GitHub Copilot ja ka ChatGPT koodi kirjutamisel.

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```
