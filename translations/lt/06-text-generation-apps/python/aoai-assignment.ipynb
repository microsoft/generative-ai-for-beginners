## Kas yra teksto generavimo programa?

Paprastai kurdami programą ji turi tam tikrą sąsają, pavyzdžiui:

- Komandinė. Konsolės programos yra tipinės programos, kuriose įvedate komandą ir ji atlieka užduotį. Pavyzdžiui, `git` yra komandinė programa.
- Vartotojo sąsaja (UI). Kai kurios programos turi grafinę vartotojo sąsają (GUI), kur spaudžiate mygtukus, įvedate tekstą, pasirenkate parinktis ir daugiau.

### Konsolės ir UI programos yra ribotos

Palyginkite su komandinėmis programomis, kuriose įvedate komandą:

- **Jie yra riboti**. Negalite įvesti bet kokios komandos, tik tų, kurias programa palaiko.
- **Kalbai specifiniai**. Kai kurios programos palaiko daug kalbų, bet pagal numatytuosius nustatymus programa sukurta konkrečiai kalbai, nors galite pridėti ir daugiau kalbų palaikymo.

### Teksto generavimo programų privalumai

Tai kuo teksto generavimo programa skiriasi?

Teksto generavimo programoje turite daugiau lankstumo, nesate apriboti tam tikrų komandų rinkiniu ar specifine įvesties kalba. Vietoj to galite naudoti natūralią kalbą bendrauti su programa. Dar vienas privalumas yra tas, kad bendraujate su duomenų šaltiniu, kuris yra apmokytas didelėje informacijos korpuse, tuo tarpu tradicinė programa gali būti ribota tuo, kas yra duomenų bazėje.

### Ką galiu sukurti su teksto generavimo programa?

Yra daug dalykų, kuriuos galite sukurti. Pavyzdžiui:

- **Pokalbį vedančią programą (chatbot)**. Chatbot'as, atsakantis į klausimus apie temas, pavyzdžiui, jūsų įmonę ir jos produktus, būtų puikus pasirinkimas.
- **Pagalbininką**. Dideli kalbos modeliai puikūs dalykams kaip teksto santrauka, išvadų iš teksto gavimas, CV ir kitų tekstų kūrimas.
- **Kodo asistentą**. Priklausomai nuo kalbos modelio, kurį naudojate, galite sukurti kodo asistentą, kuris padeda rašyti kodą. Pavyzdžiui, galite naudoti produktus kaip GitHub Copilot bei ChatGPT, kad padėtų rašyti kodą.

## Teksto generavimas

Teksto generavimui naudojame `responses.create` metodą. Štai pavyzdys:

```python

```

Aukščiau pateiktame kode sukuriame atsakymo objektą, nurodome modelį, kurį norime naudoti, ir užklausą. Tada atspausdiname sugeneruotą tekstą.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
