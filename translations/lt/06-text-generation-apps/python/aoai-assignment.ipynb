## Kas yra teksto generavimo programa?

Paprastai kuriant programą, ji turi tam tikrą sąsają, tokią kaip ši:

- Komandų pagrindu. Konsolės programos yra tipinės programos, kuriose įvedate komandą ir ji atlieka užduotį. Pavyzdžiui, `git` yra komandų pagrindu veikianti programa.
- Vartotojo sąsaja (UI). Kai kurios programos turi grafinę vartotojo sąsają (GUI), kur spustelite mygtukus, įvedate tekstą, pasirenkate parinktis ir pan.

### Konsolės ir UI programos yra ribotos

Palyginkite su komandų pagrindo programa, kurioje įvedate komandą:

- **Jos yra ribotos**. Negalite įvesti bet kokios komandos, tik tų, kurias programa palaiko.
- **Kalbai specifinės**. Kai kurios programos palaiko daug kalbų, tačiau pagal numatytuosius nustatymus programa kuriama konkrečiai kalbai, net jei galite pridėti palaikymo kitoms kalboms.

### Teksto generavimo programų privalumai

Tai kuo teksto generavimo programa skiriasi?

Teksto generavimo programoje turite daugiau lankstumo, nesate apriboti nustatytų komandų rinkiniu ar konkrečia įvesties kalba. Vietoje to galite naudoti natūralią kalbą bendraujant su programa. Kitas privalumas tas, kad jūs jau bendraujate su duomenų šaltiniu, kuris buvo apmokytas naudojant milžinišką informacijos korpusą, o tradicinė programa gali būti ribota tuo, kas yra duomenų bazėje.

### Ką galiu sukurti su teksto generavimo programa?

Yra daug dalykų, kuriuos galite sukurti. Pavyzdžiui:

- **Pokalbį vedantį robotą (chatbot)**. Pokalbių robotas, atsakantis į klausimus apie temas, pavyzdžiui, jūsų įmonę ir jos produktus, galėtų būti puikus pasirinkimas.
- **Pagalbininką**. Dideli kalbos modeliai (LLM) puikiai apibendrina tekstą, gautų įžvalgų iš teksto, generuoja tekstą, pavyzdžiui, CV ir daugiau.
- **Kodo asistentą**. Priklausomai nuo naudojamo kalbos modelio, galite sukurti kodo asistentą, kuris padeda rašyti kodą. Pavyzdžiui, galite naudoti produktus kaip GitHub Copilot arba ChatGPT, kurie padeda rašyti kodą.

## Teksto generavimas

Teksto generavimo būdas yra naudoti `responses.create` metodą. Štai pavyzdys:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Aukščiau esančiame kode mes sukuriame atsakymo objektą ir perduodame modelį, kurį norime naudoti, bei užklausą. Tada atspausdiname sugeneruotą tekstą.

## Užduotis – sukurkite receptų generatorių

Dabar, kai peržiūrėjome scenarijų, parašykime kodą, atitinkantį parodytą scenarijų. Norėdami tai padaryti, atlikite šiuos veiksmus:

1. Suraskite `prompt` kintamąjį ir pakeiskite jo kodą į šį: 


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

To further improve it, we want to add the following:

- **Filter out ingredients**. We want to be able to filter out ingredients we don't like or are allergic to. To accomplish this change, we can edit our existing prompt and add a filter condition to the end of it like so:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Above, we add `{filter}` to the end of the prompt and we also capture the filter value from the user.

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.
