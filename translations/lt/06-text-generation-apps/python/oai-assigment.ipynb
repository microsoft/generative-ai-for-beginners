## Kas yra teksto generavimo programa?

Paprastai kuriant programą ji turi tam tikrą sąsają, pavyzdžiui:

- Komandinė sąsaja. Konsolės programos yra tipinės programos, kuriose įvedate komandą ir ji atlieka užduotį. Pavyzdžiui, `git` yra programinė įranga su komandinėmis komandomis.
- Vartotojo sąsaja (UI). Kai kurios programos turi grafinę vartotojo sąsają (GUI), kur spustelite mygtukus, įvedate tekstą, pasirenkate parinktis ir dar daugiau.

### Konsolės ir UI programos turi ribotumus

Palyginkite su programa su komandine sąsaja, kur įvedate komandą:

- **Ji ribota**. Negalite įvesti bet kokios komandos, tik tas, kurias programa palaiko.
- **Kalbos specifinė**. Kai kurios programos palaiko daugelį kalbų, bet pagal nutylėjimą programa sukurta konkrečiai kalbai, nors galima pridėti daugiau kalbų palaikymo.

### Teksto generavimo programų privalumai

Tai kuo gi skiriasi teksto generavimo programa?

Teksto generavimo programoje turite daugiau lankstumo, nesate ribojamas komandų rinkiniu ar specifine įvesties kalba. Vietoje to, galite naudoti natūralią kalbą sąveikai su programa. Kita nauda – jūs jau sąveikaujate su duomenų šaltiniu, kuris buvo apmokytas didžiuliu informacijos kiekiu, o tradicinė programa gali būti ribota, kas yra duomenų bazėje.

### Ką galiu sukurti teksto generavimo programa?

Galima sukurti daug ką. Pavyzdžiui:

- **Pokalbių robotą**. Pokalbių robotas, atsakantis į klausimus apie temas kaip jūsų įmonė ir jos produktai, galėtų būti tinkamas.
- **Pagalbą**. LLM puikiai tinka tokiems dalykams kaip teksto santraukos, įžvalgų gavimas iš teksto, tekstų kūrimas, pvz., CV ir kt.
- **Kodo asistentą**. Priklausomai nuo naudojamo kalbos modelio, galite sukurti kodo asistentą, kuris padeda rašyti kodą. Pavyzdžiui, galite naudoti produktus kaip GitHub Copilot ar ChatGPT, kad padėtų rašyti kodą.

## Teksto generavimas

Tekstą generuoti galite naudodami metodą `responses.create`. Štai pavyzdys:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Aukščiau kuriame atsakymo objektą, perduodame norimą naudoti modelį ir užklausą. Tada išspausdiname sugeneruotą tekstą.

### Pokalbių atsakymai

Iki šiol matėte, kaip galime generuoti tekstą iš vienos užklausos. Responses API taip pat puikiai tinka pokalbių robotams – pateikiate žinučių sąrašą kaip `input`, kad sukurtumėte pokalbį. Štai pavyzdys:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Daugiau apie šią funkciją kitame skyriuje.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # atspausdinti atsakymą
    print("Shopping list:")
    print(response.output_text)
    ```

    Atkreipkite dėmesį į šiuos dalykus:

    - Kuriame naują užklausą pridėdami pirmos užklausos rezultatą prie naujos užklausos:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Atliekame naują užklausą, taip pat atsižvelgdami į tai, kiek tokenų paprašėme pirmoje užklausoje, šį kartą nustatome `max_output_tokens` reikšmę 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

- **Žodis apie tokenų ilgį**. Turėtume apsvarstyti, kiek tokenų mums reikia sugeneruoti norimam tekstui. Tokenai kainuoja pinigus, todėl, kai įmanoma, turėtume būti taupūs naudojant tokenų skaičių. Pvz., ar galime formuluoti užklausą taip, kad naudotume mažiau tokenų?

   Tokenų skaičių galite keisti naudodami parametrą `max_output_tokens`. Pavyzdžiui, jei norite naudoti 100 tokenų, darykite taip:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Eksperimentavimas su temperatūra**. Temperatūra yra svarbus parametras, apie kurį dar nekalbėjome, tačiau jis veikia mūsų programos veikimą. Kuo didesnė temperatūros reikšmė, tuo atsitiktinesnė bus išvestis. Atvirkščiai, kuo mažesnė temperatūros reikšmė, tuo labiau prognozuojama bus išvestis. Apsvarstykite, ar norite, kad jūsų išvestis būtų įvairi, ar ne.

   Temperatūrą galima keisti naudojant parametrą `temperature`. Pavyzdžiui, jei norite nustatyti temperatūrą 0.5, darykite taip:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```
