## Kas yra teksto generavimo programa?

Paprastai kuriant programą ji turi tam tikrą sąsają, pavyzdžiui, tokią:

- Komandų pagrindu. Konsolinės programos yra tokios, kuriose įvedate komandą ir ji atlieka užduotį. Pavyzdžiui, `git` yra komandų pagrindu veikianti programa.
- Vartotojo sąsaja (UI). Kai kurios programos turi grafinę vartotojo sąsają (GUI), kur galite spustelėti mygtukus, įvesti tekstą, pasirinkti parinktis ir pan.

### Konsolinės ir UI programos yra ribotos

Palyginkite su komandine programa, kurioje rašote komandą:

- **Ji yra ribota**. Negalite tiesiog įvesti bet kokios komandos, tik tas, kurias programa palaiko.

### Teksto generavimo programų privalumai

Kaip teksto generavimo programa skiriasi?

Teksto generavimo programoje turite daugiau lankstumo, nesate riboti nustatytų komandų ar konkrečios įvesties kalbos. Vietoj to galite naudoti natūralią kalbą bendrauti su programa. Kita nauda ta, kad jau bendraujate su duomenų šaltiniu, kuris buvo apmokytas pagal didžiulį informacijos korpusą, tuo tarpu tradicinė programa gali būti ribota tuo, kas yra duomenų bazėje.

### Ką galite sukurti naudodami teksto generavimo programą?

Yra daugybė dalykų, kuriuos galite sukurti. Pavyzdžiui:

- **Pokalbį vedantį robotą (chatbot)**. Pokalbių robotas, atsakantis į klausimus apie temas, pvz., jūsų įmonę ir jos produktus, gali būti puikus sprendimas.
- **Pagalbininką**. Didelės kalbos modeliai (LLMs) puikiai susidoroja su tokiomis užduotimis kaip teksto santraukos sudarymas, įžvalgų gavimas iš teksto, tekstų generavimas, kaip CV ir kitais dalykais.
- **Kodo asistentą**. Priklausomai nuo naudojamo kalbos modelio, galite sukurti kodą rašyti padedantį asistentą. Pavyzdžiui, galite naudoti tokius produktus kaip GitHub Copilot arba ChatGPT, kad padėtų rašyti kodą.

## Sugeneruokite tekstą

Tekstą sugeneruoti galima naudojant `responses.create` metodą. Štai pavyzdys:


Aukščiau pateiktame kode sukuriame atsakymo objektą, perduodame modelį, kurį norime naudoti, ir užklausą. Tada atspausdiname sugeneruotą tekstą.

### Pokalbių atsakymai

Iki šiol matėte, kaip galime generuoti tekstą iš vienos užklausos. Responses API taip pat puikiai tinka pokalbių robotams – pateikiate žinučių sąrašą kaip `input`, kad sudarytumėte pokalbį. Štai pavyzdys, kaip jį naudoti:

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Puiku, pažiūrėkime, kaip galime patobulinti dalykus. Norint pagerinti dalykus, norime užtikrinti, kad kodas būtų lankstus, kad ingredientai ir receptų skaičius galėtų būti patobulinti ir pakeisti.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Mes siunčiame naują užklausą, tačiau taip pat atsižvelgdami į pirmojo kvietimo paklaustų žetonų kiekį, todėl šį kartą sakome, kad `max_output_tokens` yra 1200.


- **Žodis apie žetonų ilgį**. Turėtume apsvarstyti, kiek žetonų reikia sugeneruoti norimą tekstą. Žetonai kainuoja, todėl kur įmanoma, turėtume stengtis būti taupūs žetonų skaičiumi. Pavyzdžiui, ar galime komandą suformuluoti taip, kad naudotume mažiau žetonų?

   Norėdami pakeisti naudojamų žetonų skaičių, galite naudoti parametrą `max_output_tokens`. Pavyzdžiui, jei norite naudoti 100 žetonų, tai darytumėte taip:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```
