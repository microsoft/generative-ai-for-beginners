# टेक्स्ट जनरेशन ऐप्स बनाएं

अब तक आपने इस पाठ्यक्रम में देखा है कि कुछ मुख्य कॉन्सेप्ट्स होते हैं जैसे प्रॉम्प्ट्स और एक पूरी विधा है जिसे "प्रॉम्प्ट इंजीनियरिंग" कहा जाता है। कई टूल्स जैसे ChatGPT, Office 365, Microsoft Power Platform आदि, आपको प्रॉम्प्ट्स का उपयोग करके कुछ हासिल करने में मदद करते हैं।

अगर आप ऐसी कोई सुविधा अपनी ऐप में जोड़ना चाहते हैं, तो आपको प्रॉम्प्ट्स, कम्प्लीशन जैसे कॉन्सेप्ट्स समझने होंगे और काम करने के लिए कोई लाइब्रेरी चुननी होगी। यही आप इस चैप्टर में सीखेंगे।

## परिचय

इस चैप्टर में, आप:

- openai लाइब्रेरी और उसके मुख्य कॉन्सेप्ट्स के बारे में जानेंगे।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएंगे।
- प्रॉम्प्ट, टेम्परेचर और टोकन्स जैसे कॉन्सेप्ट्स का उपयोग करके टेक्स्ट जनरेशन ऐप बनाना समझेंगे।

## लर्निंग गोल्स

इस पाठ के अंत में, आप सक्षम होंगे:

- समझा पाएंगे कि टेक्स्ट जनरेशन ऐप क्या है।
- openai का उपयोग करके टेक्स्ट जनरेशन ऐप बना पाएंगे।
- अपनी ऐप को ज्यादा या कम टोकन्स इस्तेमाल करने के लिए और आउटपुट में विविधता लाने के लिए टेम्परेचर बदलना भी सीखेंगे।

## टेक्स्ट जनरेशन ऐप क्या है?

आमतौर पर जब आप कोई ऐप बनाते हैं तो उसमें कुछ इस तरह का इंटरफेस होता है:

- कमांड-बेस्ड। कंसोल ऐप्स आमतौर पर वे ऐप्स होती हैं जिनमें आप कोई कमांड टाइप करते हैं और वह कोई काम करती है। उदाहरण के लिए, `git` एक कमांड-बेस्ड ऐप है।
- यूज़र इंटरफेस (UI)। कुछ ऐप्स में ग्राफिकल यूज़र इंटरफेस (GUI) होता है जिसमें आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं आदि।

### कंसोल और UI ऐप्स की सीमाएँ

इसे एक कमांड-बेस्ड ऐप से तुलना करें जिसमें आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड नहीं टाइप कर सकते, केवल वही जो ऐप सपोर्ट करती है।
- **भाषा-विशिष्ट**। कुछ ऐप्स कई भाषाओं को सपोर्ट करती हैं, लेकिन डिफ़ॉल्ट रूप से ऐप एक खास भाषा के लिए बनी होती है, भले ही आप और भाषाएँ जोड़ सकते हैं।

### टेक्स्ट जनरेशन ऐप्स के फायदे

तो टेक्स्ट जनरेशन ऐप किस तरह अलग है?

टेक्स्ट जनरेशन ऐप में आपके पास ज्यादा लचीलापन होता है, आप केवल कुछ कमांड्स या किसी खास इनपुट भाषा तक सीमित नहीं रहते। इसके बजाय, आप नेचुरल लैंग्वेज का उपयोग करके ऐप से संवाद कर सकते हैं। एक और फायदा यह है कि आप ऐसे डेटा स्रोत से संवाद कर रहे हैं जिसे बहुत बड़े डाटा पर प्रशिक्षित किया गया है, जबकि पारंपरिक ऐप्स केवल डेटाबेस में मौजूद जानकारी तक सीमित हो सकती हैं।

### टेक्स्ट जनरेशन ऐप से क्या-क्या बना सकते हैं?

आप बहुत कुछ बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। कोई चैटबॉट जो आपकी कंपनी और उसके प्रोडक्ट्स जैसे विषयों पर सवालों के जवाब दे सकता है।
- **हेल्पर**। LLMs टेक्स्ट को संक्षिप्त करने, टेक्स्ट से इनसाइट्स निकालने, रिज़्यूमे जैसे टेक्स्ट बनाने आदि में बहुत अच्छे हैं।
- **कोड असिस्टेंट**। आप जिस भाषा मॉडल का उपयोग करते हैं, उसके आधार पर आप एक कोड असिस्टेंट बना सकते हैं जो कोड लिखने में मदद करे। उदाहरण के लिए, आप GitHub Copilot या ChatGPT जैसे प्रोडक्ट्स का उपयोग कोड लिखने में मदद के लिए कर सकते हैं।

## कैसे शुरू करें?

आपको LLM के साथ इंटीग्रेट करने का कोई तरीका ढूंढना होगा, आमतौर पर इसके दो तरीके होते हैं:

- API का उपयोग करें। इसमें आप अपनी प्रॉम्प्ट के साथ वेब रिक्वेस्ट बनाते हैं और जनरेटेड टेक्स्ट वापस पाते हैं।
- लाइब्रेरी का उपयोग करें। लाइब्रेरीज़ API कॉल्स को आसान बनाती हैं और उपयोग में सरल बनाती हैं।

## लाइब्रेरी/SDKs

LLMs के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरीज़ हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करना और प्रॉम्प्ट भेजना आसान बनाती है।

फिर कुछ लाइब्रेरीज़ हैं जो उच्च स्तर पर काम करती हैं जैसे:

- **Langchain**। Langchain काफी प्रसिद्ध है और Python को सपोर्ट करता है।
- **Semantic Kernel**। Semantic Kernel Microsoft की लाइब्रेरी है जो C#, Python और Java भाषाओं को सपोर्ट करती है।

### API की और एंडपॉइंट ढूंढें

अब आपको अपनी `openai` लाइब्रेरी को बताना होगा कि कौन सी API की इस्तेमाल करनी है। अपनी API की ढूंढने के लिए, Azure OpenAI रिसोर्स के "Keys and Endpoint" सेक्शन में जाएं और "Key 1" की वैल्यू कॉपी करें।

  ![Azure Portal में Keys and Endpoint रिसोर्स ब्लेड](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

अब जब आपने यह जानकारी कॉपी कर ली है, तो लाइब्रेरीज़ को इसका उपयोग करने के लिए निर्देश दें।

> [!NOTE]
> अपनी API की कोड से अलग रखना अच्छा रहता है। आप ऐसा environment variables का उपयोग करके कर सकते हैं।
> - अपनी .env फाइल में environment variable `AZURE_OPENAI_API_KEY` को अपनी API की से सेट करें। अगर आपने इस कोर्स के पिछले अभ्यास पूरे कर लिए हैं, तो आप पहले से तैयार हैं।


### Azure के लिए कॉन्फ़िगरेशन सेटअप करें

अगर आप Azure OpenAI का उपयोग कर रहे हैं, तो कॉन्फ़िगरेशन ऐसे सेट करें:

```python
client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT')
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ऊपर हम ये सेट कर रहे हैं:

- `api_key`, यह आपकी API की है जो Azure Portal में मिलती है।
- `api_version`, यह उस API का वर्शन है जिसे आप उपयोग करना चाहते हैं। इस लेखन के समय, लेटेस्ट वर्शन `2023-10-01-preview` है।
- `azure_endpoint`, यह API का एंडपॉइंट है। आप इसे Azure Portal में अपनी API की के पास देख सकते हैं।

> [!NOTE]
> `os.environ` एक फंक्शन है जो environment variables पढ़ता है। आप इसका उपयोग `AZURE_OPENAI_API_KEY` और `AZURE_OPENAI_ENDPOINT` जैसे environment variables पढ़ने के लिए कर सकते हैं।

## टेक्स्ट जनरेट करें

टेक्स्ट जनरेट करने का तरीका है `chat.completion` क्लास का उपयोग करना। उदाहरण देखें:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

ऊपर दिए गए कोड में, हम एक कम्प्लीशन ऑब्जेक्ट बनाते हैं और उसमें मॉडल और प्रॉम्प्ट पास करते हैं। फिर हम जनरेटेड टेक्स्ट प्रिंट करते हैं।

### चैट कम्प्लीशन

अब तक आपने देखा कि हम टेक्स्ट जनरेट करने के लिए `Completion` का उपयोग कर रहे थे। लेकिन एक और क्लास है `ChatCompletion` जो चैटबॉट्स के लिए ज्यादा उपयुक्त है। इसका उपयोग करने का उदाहरण:

```python
client = AzureOpenAI(
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT'), 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-05-15"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

इस फंक्शनलिटी के बारे में आगे के चैप्टर में और जानेंगे।

## अभ्यास - अपनी पहली टेक्स्ट जनरेशन ऐप

अब जब हमने Azure OpenAI सर्विस को सेटअप और कॉन्फ़िगर करना सीख लिया है, तो अब अपनी पहली टेक्स्ट जनरेशन ऐप बनाने का समय है। अपनी ऐप बनाने के लिए ये स्टेप्स फॉलो करें:


1. एक वर्चुअल एनवायरनमेंट बनाएं और openai इंस्टॉल करें:

  > [!NOTE] यह स्टेप जरूरी नहीं है अगर आप यह नोटबुक Codespaces या Devcontainer में चला रहे हैं


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> अगर आप Windows इस्तेमाल कर रहे हैं तो `source venv/bin/activate` की जगह `venv\Scripts\activate` टाइप करें।

> [!NOTE]
> अपना Azure OpenAI कुंजी खोजने के लिए https://portal.azure.com/ पर जाएं, वहां `Open AI` सर्च करें, फिर `Open AI resource` चुनें, उसके बाद `Keys and Endpoint` चुनें और `Key 1` का मान कॉपी करें।


1. एक *app.py* फ़ाइल बनाएं और उसमें निम्नलिखित कोड दें:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## अलग-अलग चीज़ों के लिए अलग-अलग प्रकार के प्रॉम्प्ट्स

अब आपने देखा कि प्रॉम्प्ट का इस्तेमाल करके टेक्स्ट कैसे जनरेट किया जा सकता है। आपके पास एक प्रोग्राम भी है जिसे आप बदल सकते हैं और अलग-अलग तरह का टेक्स्ट जनरेट कर सकते हैं।

प्रॉम्प्ट्स का इस्तेमाल कई तरह के कामों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी खास तरह का टेक्स्ट जनरेट करना**। जैसे, आप एक कविता, क्विज़ के सवाल आदि जनरेट कर सकते हैं।
- **जानकारी ढूंढना**। आप प्रॉम्प्ट्स का इस्तेमाल जानकारी खोजने के लिए कर सकते हैं, जैसे 'वेब डेवलपमेंट में CORS का क्या मतलब है?'।
- **कोड जनरेट करना**। आप प्रॉम्प्ट्स का इस्तेमाल कोड जनरेट करने के लिए कर सकते हैं, जैसे ईमेल वैलिडेट करने के लिए रेगुलर एक्सप्रेशन बनाना या फिर पूरा प्रोग्राम जनरेट करना, जैसे एक वेब ऐप।

## एक व्यावहारिक उदाहरण: रेसिपी जनरेटर

> "मुझे इन सामग्रियों के साथ कोई डिश बनाने की 5 रेसिपी दिखाओ: चिकन, आलू और गाजर। हर रेसिपी में इस्तेमाल की गई सारी सामग्री की लिस्ट दो।"

ऊपर दिए गए प्रॉम्प्ट के आधार पर, आपको कुछ ऐसा जवाब मिल सकता है:

अब जब हमने एक परिदृश्य को देखा है, तो चलिए उस परिदृश्य से मेल खाने वाला कोड लिखते हैं। ऐसा करने के लिए, निम्नलिखित चरणों का पालन करें:

1. मौजूदा *app.py* फ़ाइल को शुरुआती बिंदु के रूप में उपयोग करें
1. `prompt` वेरिएबल को ढूंढें और उसके कोड को निम्नलिखित में बदलें:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

बहुत बढ़िया, अब देखते हैं कि हम चीजों को कैसे बेहतर बना सकते हैं। चीजों को बेहतर बनाने के लिए, हम चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या को आसानी से बदला और सुधारा जा सके।
