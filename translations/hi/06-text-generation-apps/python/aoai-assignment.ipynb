# टेक्स्ट जनरेशन ऐप बनाएं

आपने अब तक इस पाठ्यक्रम के माध्यम से देखा है कि कुछ मूल अवधारणाएं हैं जैसे प्रॉम्प्ट और यहां तक कि एक पूरी विशेषज्ञता जिसे "प्रॉम्प्ट इंजीनियरिंग" कहा जाता है। कई टूल जिनसे आप बातचीत कर सकते हैं जैसे ChatGPT, Office 365, Microsoft Power Platform और अन्य, प्रॉम्प्ट का उपयोग करके आपको कुछ हासिल करने में सहायता करते हैं।

आपके लिए ऐसी अनुभव को एक ऐप में जोड़ने के लिए, आपको प्रॉम्प्ट, पूर्णता जैसी अवधारणाओं को समझने और काम करने के लिए एक लाइब्रेरी चुनने की आवश्यकता होती है। यही आप इस अध्याय में सीखेंगे।

## परिचय

इस अध्याय में, आप:

- openai लाइब्रेरी और उसकी मूल अवधारणाओं के बारे में सीखेंगे।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएंगे।
- टेक्स्ट जनरेशन ऐप बनाने के लिए प्रॉम्प्ट, तापमान (temperature), और टोकन जैसी अवधारणाओं का उपयोग कैसे करें यह समझेंगे।

## सीखने के लक्ष्य

इस पाठ के अंत में, आप सक्षम होंगे:

- टेक्स्ट जनरेशन ऐप क्या है यह समझाना।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाना।
- अपने ऐप को अधिक या कम टोकन उपयोग करने के लिए और आउटपुट को विविध बनाने के लिए तापमान बदलने के लिए कॉन्फ़िगर करना।

## टेक्स्ट जनरेशन ऐप क्या है?

आमतौर पर जब आप एक ऐप बनाते हैं तो उसमें कुछ तरह का इंटरफेस होता है जैसे निम्नलिखित:

- कमांड-आधारित। कंसोल ऐप सामान्य ऐप होते हैं जहां आप कमांड टाइप करते हैं और यह कार्य करता है। उदाहरण के लिए, `git` एक कमांड-आधारित ऐप है।
- उपयोगकर्ता इंटरफ़ेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUI) होते हैं जहां आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं आदि।

### कंसोल और UI ऐप्स सीमित हैं

इसके बजाय एक कमांड-आधारित ऐप की तुलना करें जहां आप एक कमांड टाइप करते हैं:

- **यह सीमित है**। आप किसी भी कमांड को टाइप नहीं कर सकते हैं, केवल वे कमांड जो ऐप का समर्थन करता है।
- **भाषा विशिष्ट**। कुछ ऐप कई भाषाओं का समर्थन करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप एक विशेष भाषा के लिए बनाया जाता है, भले ही आप अधिक भाषा समर्थन जोड़ सकें।

### टेक्स्ट जनरेशन ऐप के लाभ

तो टेक्स्ट जनरेशन ऐप कैसे अलग है?

एक टेक्स्ट जनरेशन ऐप में, आपके पास अधिक लचीलापन होता है, आप कमांड सेट या एक विशिष्ट इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप प्राकृतिक भाषा का उपयोग करके ऐप के साथ बातचीत कर सकते हैं। एक और लाभ यह है कि आप पहले से ही एक डेटा स्रोत के साथ इंटरैक्ट कर रहे होते हैं जिसे व्यापक जानकारी के एक विशाल कॉर्पस पर प्रशिक्षित किया गया है, जबकि एक पारंपरिक ऐप अक्सर डेटाबेस में मौजूद जानकारी तक ही सीमित हो सकता है।

### मैं टेक्स्ट जनरेशन ऐप के साथ क्या बना सकता हूँ?

आप कई चीजें बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। एक चैटबॉट जो आपके कंपनी और उसके उत्पादों के बारे में प्रश्नों के उत्तर देता है, एक अच्छा विकल्प हो सकता है।
- **मददगार**। LLMs टेक्स्ट का सारांश बनाने, टेक्स्ट से अंतर्दृष्टि प्राप्त करने, जैसे रिज्यूमे और अन्य टेक्स्ट उत्पन्न करने में माहिर होते हैं।
- **कोड सहायक**। जिस भाषा मॉडल का आप उपयोग करते हैं, उसके आधार पर आप एक कोड सहायक बना सकते हैं जो कोड लिखने में मदद करता है। उदाहरण के लिए, आप GitHub Copilot और ChatGPT जैसे उत्पादों का उपयोग करके कोड लिखने में सहायता प्राप्त कर सकते हैं।

## मैं कैसे शुरू कर सकता हूँ?

खैर, आपको एक LLM के साथ एकीकृत होने का तरीका खोजना होगा जो सामान्यतः निम्न दो दृष्टिकोणों को शामिल करता है:

- API का उपयोग करें। यहां आप वेब अनुरोधों का निर्माण करते हैं जिसमें आपका प्रॉम्प्ट शामिल होता है और आपको उत्पन्न टेक्स्ट वापस मिलता है।
- लाइब्रेरी का उपयोग करें। लाइब्रेरी API कॉल्स को संलग्न करती हैं और उनका उपयोग आसान बनाती हैं।

## लाइब्रेरी/SDKs

LLMs के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरियां हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से जुड़ना और प्रॉम्प्ट भेजना आसान बनाती है।

फिर कुछ उच्च स्तरीय लाइब्रेरी भी हैं जैसे:

- **Langchain**। Langchain प्रसिद्ध है और Python का समर्थन करता है।
- **Semantic Kernel**। Semantic Kernel एक Microsoft की लाइब्रेरी है जो C#, Python, और Java भाषाओं का समर्थन करती है।

### API कुंजी और एंडपॉइंट पता लगाएं

इस बिंदु पर, आपको अपनी `openai` लाइब्रेरी को यह बताना होगा कि कौन-सी API कुंजी का उपयोग करना है। अपनी API कुंजी खोजने के लिए, अपने Azure OpenAI संसाधन के "Keys and Endpoint" सेक्शन पर जाएं और "Key 1" मान कॉपी करें।

  ![Azure Portal में Keys and Endpoint संसाधन ब्लेड](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

अब जब आपके पास यह जानकारी कॉपी हो गई है, तो चलिए लाइब्रेरियों को इसे उपयोग करने के लिए निर्देशित करें।

> [!NOTE]
> यह आपके कोड से अपनी API कुंजी को अलग करने लायक है। आप ऐसा वातावरण चर (environment variables) का उपयोग करके कर सकते हैं।
> - अपने .env फ़ाइल में API कुंजी के लिए `AZURE_OPENAI_API_KEY` नाम का वातावरण चर सेट करें। यदि आपने इस पाठ्यक्रम के पिछले अभ्यास पूरे किए हैं, तो आप पूरी तरह तैयार हैं।


### Azure कॉन्फ़िगरेशन सेटअप करें

यदि आप Azure OpenAI का उपयोग कर रहे हैं, तो यहां कॉन्फ़िगरेशन सेटअप करने का तरीका है। Responses API Azure OpenAI के **v1 endpoint** से सेवा प्राप्त करता है, इसलिए हम `OpenAI` क्लाइंट को `<your-endpoint>/openai/v1/` की ओर इंगित करते हैं:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ऊपर हम निम्न सेट कर रहे हैं:

- `api_key`, यह आपकी Azure Portal में मिली API कुंजी है।
- `base_url`, यह आपका Azure OpenAI एंडपॉइंट है जिसके बाद `/openai/v1/` जुड़ा हुआ है। आप एंडपॉइंट Azure Portal में अपनी API कुंजी के बगल में पा सकते हैं। v1 एंडपॉइंट का उपयोग करने का मतलब है कि अब आपको `api_version` पास करने की जरूरत नहीं है।
- `deployment`, यह उस मॉडल डिप्लॉयमेंट का नाम है जिसे आपने Foundry पोर्टल में बनाया है।

> [!NOTE]
> `os.environ` एक मैपिंग है जो वातावरण चर पढ़ता है। आप इसका उपयोग `AZURE_OPENAI_API_KEY` और `AZURE_OPENAI_ENDPOINT` जैसे वातावरण चर पढ़ने के लिए कर सकते हैं।

## टेक्स्ट जनरेट करें

टेक्स्ट जनरेट करने का तरीका `responses.create` मेथड का उपयोग करना है। यहां एक उदाहरण है:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ऊपर के कोड में, हम एक प्रतिक्रिया वस्तु बनाते हैं और मॉडल सहित प्रॉम्प्ट पास करते हैं। फिर हम उत्पन्न टेक्स्ट प्रिंट करते हैं।

### चैट प्रतिक्रियाएँ

Responses API एकल-टर्न टेक्स्ट जनरेशन और बहु-टर्न चैटबॉट दोनों के लिए उपयुक्त है - आप सिर्फ बातचीत बनाने के लिए `input` सूची में अधिक संदेश प्रदान करते हैं:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

इस कार्यक्षमता पर अधिक जानकारी अगले अध्याय में मिलेगी।

## अभ्यास - आपका पहला टेक्स्ट जनरेशन ऐप

अब जब हमने Azure OpenAI सेवा को सेटअप और कॉन्फ़िगर करना सीख लिया है, तो आपके पहला टेक्स्ट जनरेशन ऐप बनाने का समय है। अपने ऐप को बनाने के लिए निम्न कदमों का पालन करें:


1. एक वर्चुअल एन्वायरनमेंट बनाएं और openai इंस्टॉल करें:

  > [!NOTE] यदि आप यह नोटबुक Codespaces पर या Devcontainer के अंदर चला रहे हैं तो यह कदम आवश्यक नहीं है


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> यदि आप विंडोज़ का उपयोग कर रहे हैं तो `source venv/bin/activate` की जगह `venv\Scripts\activate` टाइप करें।

> [!NOTE]
> अपनी Azure OpenAI कुंजी खोजने के लिए https://portal.azure.com/ पर जाएं और `Open AI` खोजें, फिर `Open AI resource` चुनें और उसके बाद `Keys and Endpoint` चुनें और `Key 1` का मान कॉपी करें।


1. एक *app.py* फाइल बनाएं और इसे निम्नलिखित कोड दें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## अलग-अलग चीज़ों के लिए अलग-अलग प्रकार के प्रॉम्प्ट

अब आपने देखा कि प्रॉम्प्ट का उपयोग करके टेक्स्ट कैसे जनरेट किया जाता है। आपके पास एक प्रोग्राम भी है जो चल रहा है जिसे आप संशोधित और बदलकर विभिन्न प्रकार के टेक्स्ट जनरेट कर सकते हैं।

प्रॉम्प्ट्स का उपयोग सभी प्रकार के कार्यों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी प्रकार का टेक्स्ट जनरेट करना**। उदाहरण के लिए, आप एक कविता, क्विज़ के लिए प्रश्न आदि जेनरेट कर सकते हैं।
- **जानकारी खोजना**। आप प्रॉम्प्ट्स का उपयोग जानकारी खोजने के लिए कर सकते हैं जैसे इस उदाहरण में 'वेब डेवलपमेंट में CORS का क्या मतलब है?'

अब जब हमने एक परिदृश्य खेल लिया है, तो चलिए उस प्रदर्शित परिदृश्य से मेल खाते कोड को लिखते हैं। ऐसा करने के लिए, निम्नलिखित चरणों का पालन करें:

1. `prompt` वेरिएबल को ढूंढें और उसके कोड को निम्नलिखित में बदलें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    ```python
    old_prompt_result = response.output_text

    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    ```

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
