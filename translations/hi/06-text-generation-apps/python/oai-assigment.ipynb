# टेक्स्ट जनरेशन ऐप्स बनाएं

आपने अब तक इस पाठ्यक्रम के माध्यम से देखा है कि कोर अवधारणाएं जैसे प्रॉम्प्ट और यहां तक कि "प्रॉम्प्ट इंजीनियरिंग" नामक एक संपूर्ण अनुशासन है। कई उपकरण जिनके साथ आप इंटरैक्ट कर सकते हैं जैसे ChatGPT, Office 365, Microsoft Power Platform और अन्य, आपको कुछ पूरा करने के लिए प्रॉम्प्ट का उपयोग करके समर्थन करते हैं।

अपने ऐप में ऐसा अनुभव जोड़ने के लिए, आपको प्रॉम्प्ट, पूर्णताएँ जैसी अवधारणाओं को समझना होगा और एक लाइब्रेरी चुननी होगी जिससे आप काम कर सकें। यही आप इस अध्याय में सीखेंगे।

## परिचय

इस अध्याय में, आप:

- openai लाइब्रेरी और उसकी कोर अवधारणाओं के बारे में जानेंगे।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएंगे।
- टेक्स्ट जनरेशन ऐप बनाने के लिए प्रॉम्प्ट, तापमान (temperature), और टोकन (tokens) जैसी अवधारणाओं का उपयोग कैसे करें, यह समझेंगे।

## सीखने के उद्देश्य

इस पाठ के अंत में, आप सक्षम होंगे:

- टेक्स्ट जनरेशन ऐप क्या होता है, समझाएँ।
- openai का उपयोग करके टेक्स्ट जनरेशन ऐप बनाएँ।
- अधिक या कम टोकन का उपयोग करने के लिए अपने ऐप को कॉन्फ़िगर करें और साथ ही विभिन्न आउटपुट के लिए तापमान (temperature) बदलें।

## टेक्स्ट जनरेशन ऐप क्या है?

सामान्यतः जब आप कोई ऐप बनाते हैं तो उसमें कुछ प्रकार का इंटरफ़ेस होता है जैसे निम्नलिखित:

- कमांड-आधारित। कंसोल ऐप्स ऐसे ऐप्स होते हैं जहाँ आप एक कमांड टाइप करते हैं और वह कार्य करता है। उदाहरण के लिए, `git` एक कमांड-आधारित ऐप है।
- यूजर इंटरफेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUIs) होते हैं जहाँ आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं और अधिक।

### कंसोल और UI ऐप्स सीमित होते हैं

इसे एक कमांड-आधारित ऐप से तुलना करें जहाँ आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड टाइप नहीं कर सकते, केवल वही जो ऐप समर्थन करता है।
- **भाषा विशिष्ट**। कुछ ऐप्स कई भाषाओं का समर्थन करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप किसी विशेष भाषा के लिए बनाया जाता है, भले ही आप अधिक भाषा समर्थन जोड़ सकें।

### टेक्स्ट जनरेशन ऐप्स के फायदे

तो टेक्स्ट जनरेशन ऐप अलग कैसे है?

एक टेक्स्ट जनरेशन ऐप में, आपके पास अधिक लचीलापन होता है, आप किसी सेट कमांड या किसी विशेष इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप प्राकृतिक भाषा का उपयोग करके ऐप से इंटरैक्ट कर सकते हैं। एक और फायदा यह है कि आप पहले से ही ऐसे डेटा स्रोत के साथ इंटरैक्ट कर रहे हैं जिस पर बहुत बड़े सूचना संग्रह (corpus) पर प्रशिक्षित किया गया है, जबकि पारंपरिक ऐप डाटाबेस में सीमित हो सकता है।

### मैं टेक्स्ट जनरेशन ऐप से क्या बना सकता हूँ?

आप कई चीजें बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। आपके कंपनी और उसके उत्पादों जैसे विषयों के बारे में प्रश्नों का उत्तर देने वाला एक चैटबॉट एक अच्छा विकल्प हो सकता है।
- **सहायक**। LLMs टेक्स्ट को सारांशित करने, टेक्स्ट से अंतर्दृष्टि प्राप्त करने, रिज्यूमे जैसे टेक्स्ट उत्पन्न करने जैसी चीजों में उत्कृष्ट हैं।
- **कोड असिस्टेंट**। उपयोग किए गए भाषा मॉडल के आधार पर, आप ऐसा कोड असिस्टेंट बना सकते हैं जो कोड लिखने में आपकी मदद करता है। उदाहरण के लिए, आप GitHub Copilot और ChatGPT जैसे उत्पादों का उपयोग कोड लिखने में सहायता के लिए कर सकते हैं।

## मैं कैसे शुरू कर सकता हूँ?

खैर, आपको LLM के साथ एकीकरण करने का तरीका खोजना होगा जिसमें आम तौर पर निम्न दो तरीके शामिल होते हैं:

- एक API का उपयोग करें। यहां आप वेब अनुरोध आपके प्रॉम्प्ट के साथ बनाते हैं और जनरेटेड टेक्स्ट वापस पाते हैं।
- एक लाइब्रेरी का उपयोग करें। लाइब्रेरी API कॉल्स को एन्कैप्सुलेट करती हैं और उपयोग करना आसान बनाती हैं।

## लाइब्रेरी/SDKs

LLM के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरियाँ हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करना और प्रॉम्प्ट भेजना आसान बनाती है।

फिर कुछ उच्च स्तरीय लाइब्रेरी भी हैं जैसे:

- **Langchain**। Langchain प्रसिद्ध है और Python का समर्थन करता है।
- **Semantic Kernel**। Semantic Kernel माइक्रोसॉफ्ट की एक लाइब्रेरी है जो C#, Python, और Java भाषाओं का समर्थन करती है।

## openai का उपयोग करके पहला ऐप

देखते हैं कि हम अपना पहला ऐप कैसे बना सकते हैं, हमें कौन-कौन सी लाइब्रेरी चाहिए, कितना आवश्यक है, और अन्य बातें।

### openai इंस्टॉल करें

  > [!NOTE] यदि आप यह नोटबुक Codespaces या Devcontainer में चला रहे हैं तो यह कदम आवश्यक नहीं है


OpenAI या Azure OpenAI के साथ इंटरैक्ट करने के लिए कई लाइब्रेरी उपलब्ध हैं। कई प्रोग्रामिंग भाषाओं का उपयोग भी किया जा सकता है जैसे C#, Python, JavaScript, Java और अधिक।  
हमने `openai` Python लाइब्रेरी का उपयोग करने का चुनाव किया है, इसलिए हम इसे `pip` के ज़रिए इंस्टॉल करेंगे।

```bash

```

यदि आप यह नोटबुक Codespaces या Dev Container में नहीं चला रहे हैं, तो आपको अपनी मशीन पर [Python](https://www.python.org/) भी इंस्टॉल करना होगा।

### एक संसाधन बनाएँ और अपना API कुंजी खोजें

यदि आपने अभी तक नहीं किया है, तो आपको निम्नलिखित कदम करने होंगे:

- OpenAI पर एक खाता बनाएं <https://platform.openai.com/signup>।
- अब अपनी API कुंजी प्राप्त करें <https://platform.openai.com/api-keys>।

>[!NOTE]
> अपने API कुंजी को अपने कोड से अलग रखना महत्वपूर्ण है। आप ऐसा पर्यावरण चर (environment variable) का उपयोग करके कर सकते हैं।
> - अपने .env फ़ाइल में पर्यावरण चर `OPENAI_KEY` अपने API कुंजी से सेट करें। यदि आपने इस कोर्स के पूर्व अभ्यास पूर्ण कर लिए हैं, तो आप तैयार हैं।
> - ध्यान दें कि API कुंजी केवल एक बार ही पहुँच योग्य होगी। इसलिए इसे सही ढंग से कॉपी किया गया है यह सुनिश्चित करना आवश्यक है। यदि यह सही से काम नहीं करता है, तो कुंजी को हटा दें और नई कुंजी बनाएँ।


### OpenAI कॉन्फ़िगरेशन सेट करें

यदि आप OpenAI का उपयोग कर रहे हैं, तो कॉन्फ़िगरेशन सेटअप करने का तरीका यहाँ है:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

ऊपर हम निम्न सेट कर रहे हैं:

- `api_key`, यह आपकी OpenAI डैशबोर्ड में पाया जाने वाला API कुंजी है।
- `deployment`, यह आपका GPT संस्करण है।

> [!NOTE]
> `os.environ` एक फ़ंक्शन है जो पर्यावरण चर पढ़ता है। आप इसका उपयोग `OPENAI_API_KEY` जैसे पर्यावरण चर पढ़ने के लिए कर सकते हैं।

## टेक्स्ट जनरेट करें

टेक्स्ट जनरेट करने का तरीका `responses.create` मेथड का उपयोग करना है। उदाहरण यहाँ है:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ऊपर के कोड में, हमने एक प्रतिक्रिया ऑब्जेक्ट बनाया और जिस मॉडल का उपयोग करना था और प्रॉम्प्ट को पास किया। फिर हमने जनरेटेड टेक्स्ट प्रिंट किया।

### चैट प्रतिक्रिया

अब तक, आपने देखा कि हम एकल प्रॉम्प्ट से टेक्स्ट कैसे जनरेट कर सकते हैं। Responses API चैटबॉट के लिए भी उपयुक्त है - आप बातचीत बनाने के लिए संदेशों की सूची `input` के रूप में देते हैं। इसका एक उदाहरण यहाँ है:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

इस कार्यक्षमता पर एक आगामी अध्याय में और अधिक जानकारी होगी।

## अभ्यास - आपका पहला टेक्स्ट जनरेशन ऐप

अब जब हमने OpenAI सेवा को सेटअप करना और कॉन्फ़िगर करना सीख लिया है, तो अपने पहले टेक्स्ट जनरेशन ऐप को बनाने का समय है। अपना ऐप बनाने के लिए, निम्नलिखित कदमों का पालन करें:


1. एक वर्चुअल वातावरण बनाएं और openai इंस्टॉल करें:

  > [!NOTE] यदि आप इस नोटबुक को Codespaces या Devcontainer में चला रहे हैं तो यह चरण आवश्यक नहीं है


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> यदि आप विंडोज़ का उपयोग कर रहे हैं तो `source venv/bin/activate` के बजाय `venv\Scripts\activate` टाइप करें। 

> [!NOTE]
> अपने OpenAI कुंजी को खोजने के लिए https://platform.openai.com/settings/organization/api-keys पर जाएं और `API keys` खोजें। आप वहां एक नई कुंजी बना सकते हैं और तुरंत उसका मान कॉपी कर सकते हैं।


1. एक *app.py* फ़ाइल बनाएं और इसमें निम्न कोड दें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    आपको निम्नलिखित के समान एक आउटपुट देखना चाहिए:

    ```output
     बहुत दुखी _____.

    एक समय की बात है एक बहुत दुखी जलपरी थी।
    ```


## विभिन्न कार्यों के लिए विभिन्न प्रकार के प्रॉम्प्ट  

अब आपने देखा कि प्रॉम्प्ट का उपयोग करके टेक्स्ट कैसे जनरेट किया जाता है। आपके पास एक प्रोग्राम भी चल रहा है जिसे आप अलग-अलग प्रकार के टेक्स्ट जनरेट करने के लिए संशोधित और बदल सकते हैं।  

प्रॉम्प्ट का उपयोग सभी तरह के कार्यों के लिए किया जा सकता है। उदाहरण के लिए:  

- **किसी प्रकार का टेक्स्ट जनरेट करना**। उदाहरण के लिए, आप एक कविता, प्रश्नोत्तरी के लिए प्रश्न आदि जनरेट कर सकते हैं।  
- **जानकारी खोजना**। आप प्रॉम्प्ट का उपयोग जानकारी खोजने के लिए कर सकते हैं, जैसे उदाहरण में 'वेब विकास में CORS का क्या मतलब है?'।  
- **कोड जनरेट करना**। आप प्रॉम्प्ट का उपयोग कोड जनरेट करने के लिए कर सकते हैं, उदाहरण के लिए ईमेल की वैधता जांचने के लिए एक रेगुलर एक्सप्रेशन विकसित करना या पूरा प्रोग्राम जनरेट करना, जैसे एक वेब ऐप?

## अभ्यास - एक रेसिपी जनरेटर बनाएं

अब जब हमने एक परिदृश्‍य का अभ्यास किया है, तो चलिए उस दिखाए गए परिदृश्‍य से मेल खाने वाला कोड लिखते हैं। ऐसा करने के लिए, निम्नलिखित कदम उठाएं:

1. मौजूदा *app.py* फ़ाइल को शुरुआती बिंदु के रूप में उपयोग करें
1. `prompt` वेरिएबल को ढूंढें और इसका कोड निम्नलिखित में बदलें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

यदि आप अब कोड चलाते हैं, तो आपको ऐसा आउटपुट दिखना चाहिए:

> नोट, आपका LLM non-deterministic है, इसलिए हर बार प्रोग्राम चलाने पर आपको अलग परिणाम मिल सकते हैं।

बढ़िया, आइए देखें कि हम चीजों को कैसे बेहतर बना सकते हैं। चीजों को बेहतर बनाने के लिए, हम यह सुनिश्चित करना चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या में सुधार और परिवर्तन किया जा सके।

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### फ़िल्टर और शॉपिंग सूची जोड़कर सुधारें

अब हमारे पास एक कार्यशील ऐप है जो रेसिपी बना सकता है और यह लचीला है क्योंकि यह उपयोगकर्ता से इनपुट पर निर्भर करता है, दोनों रेसिपी की संख्या और उपयोग की गई सामग्री पर भी।

इसे और बेहतर बनाने के लिए, हम निम्नलिखित जोड़ना चाहते हैं:

- **सामग्री फ़िल्टर करना**। हम उन सामग्रियों को फ़िल्टर कर सकें जो हमें पसंद नहीं हैं या जिनसे हम एलर्जी हैं। इस बदलाव को पूरा करने के लिए, हम अपने मौजूदा प्रॉम्प्ट को संपादित कर सकते हैं और उसके अंत में एक फ़िल्टर की स्थिति जोड़ सकते हैं इस प्रकार:

- **शॉपिंग सूची बनाएँ**। हम एक शॉपिंग सूची बनाना चाहते हैं, यह ध्यान में रखते हुए कि हमारे पास घर पर क्या है।

    इस कार्यक्षमता के लिए, हम या तो एक ही प्रॉम्प्ट में सब कुछ हल करने की कोशिश कर सकते हैं या इसे दो प्रॉम्प्ट में विभाजित कर सकते हैं। चलिए बाद के दृष्टिकोण को आज़माते हैं। यहाँ हम एक अतिरिक्त प्रॉम्प्ट जोड़ने का सुझाव दे रहे हैं, लेकिन उसके काम करने के लिए, हमें पहले प्रॉम्प्ट के परिणाम को बाद वाले प्रॉम्प्ट के संदर्भ के रूप में जोड़ना होगा।

    कोड में उस भाग को खोजें जो पहले प्रॉम्प्ट के परिणाम को प्रिंट करता है और इसके नीचे निम्नलिखित कोड जोड़ें:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # प्रतिक्रिया प्रिंट करें
    print("Shopping list:")
    print(response.output_text)
    ```

    - हम एक नया प्रॉम्प्ट बना रहे हैं जिसमें पहले प्रॉम्प्ट का परिणाम नया प्रॉम्प्ट में जोड़ा गया है:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - हम एक नई रिक्वेस्ट बनाते हैं, लेकिन पहले प्रॉम्प्ट में पूछे गए टोकन्स की संख्या को भी ध्यान में रखते हुए, इसलिए इस बार हम कहते हैं `max_output_tokens` 1200 है।

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        इस कोड को चलाने पर हमें निम्नलिखित आउटपुट मिलता है:

- **टोकन की लंबाई के बारे में एक शब्द**। हमें यह विचार करना चाहिए कि हमें टेक्स्ट उत्पन्न करने के लिए कितने टोकन चाहिए। टोकन की लागत होती है, इसलिए जहाँ संभव हो, हमें उपयोग किए गए टोकन की संख्या में आर्थिक होना चाहिए। उदाहरण के लिए, क्या हम प्रॉम्प्ट को इस तरह बना सकते हैं कि हम कम टोकन इस्तेमाल करें?

   टोकन का उपयोग बदलने के लिए, आप `max_output_tokens` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, यदि आप 100 टोकन का उपयोग करना चाहते हैं, तो आप ऐसा करेंगे:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **टेम्परेचर के साथ प्रयोग करना**। टेम्परेचर ऐसी चीज़ है जिसका हमने अब तक उल्लेख नहीं किया है, लेकिन यह हमारे प्रोग्राम के प्रदर्शन के लिए महत्वपूर्ण संदर्भ है। जितनी अधिक टेम्परेचर वैल्यू होगी, आउटपुट उतना ही अधिक यादृच्छिक होगा। इसके विपरीत, टेम्परेचर वैल्यू जितनी कम होगी, आउटपुट उतना ही अधिक पूर्वानुमेय होगा। सोचें कि क्या आप अपने आउटपुट में विविधता चाहते हैं या नहीं।

   टेम्परेचर को बदलने के लिए, आप `temperature` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, यदि आप टेम्परेचर 0.5 का उपयोग करना चाहते हैं, तो आप ऐसा करेंगे:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > ध्यान दें, 1.0 के करीब टेम्परेचर अधिक विविध आउटपुट देता है।
