# पाठ उत्पन्न करने वाले ऐप बनाएं

आपने अब तक इस पाठ्यक्रम के दौरान देखा है कि ऐसे मूलभूत अवधारणाएं हैं जैसे प्रॉम्प्ट और यहां तक कि एक पूरा अनुशासन भी है जिसे "प्रॉम्प्ट इंजीनियरिंग" कहा जाता है। कई उपकरण जिनके साथ आप इंटरैक्ट कर सकते हैं जैसे ChatGPT, Office 365, Microsoft Power Platform और अधिक, आपको कुछ पूरा करने के लिए प्रॉम्प्ट्स का उपयोग करने का समर्थन करते हैं।

यदि आप ऐसी अनुभव को किसी ऐप में जोड़ना चाहते हैं, तो आपको प्रॉम्प्ट्स, पूर्णताओं जैसी अवधारणाओं को समझना होगा और काम करने के लिए एक लाइब्रेरी चुननी होगी। यही आप इस अध्याय में सीखेंगे।

## परिचय

इस अध्याय में, आप:

- openai लाइब्रेरी और इसकी मूल अवधारणाओं के बारे में जानेंगे।
- openai का उपयोग करके एक पाठ उत्पन्न करने वाला ऐप बनाएंगे।
- प्रॉम्प्ट, तापमान और टोकन जैसी अवधारणाओं का उपयोग करके एक पाठ उत्पन्न करने वाला ऐप कैसे बनाएं, यह समझेंगे।

## सीखने के लक्ष्य

इस पाठ के अंत में, आप सक्षम होंगे:

- यह समझाना कि एक पाठ उत्पन्न करने वाला ऐप क्या है।
- openai का उपयोग करके एक पाठ उत्पन्न करने वाला ऐप बनाना।
- अपने ऐप को अधिक या कम टोकन उपयोग करने के लिए कॉन्फ़िगर करना और साथ ही परिणाम में विविधता के लिए तापमान बदलना।

## पाठ उत्पन्न करने वाला ऐप क्या है?

सामान्यतः जब आप कोई ऐप बनाते हैं तो उसमें कुछ प्रकार का इंटरफ़ेस होता है जैसे निम्नलिखित:

- कमांड आधारित। कंसोल ऐप्स वे सामान्य ऐप्स होते हैं जहां आप एक कमांड टाइप करते हैं और यह कोई कार्य करता है। उदाहरण के लिए, `git` एक कमांड-आधारित ऐप है।
- उपयोगकर्ता इंटरफ़ेस (UI)। कुछ ऐप्स में ग्राफ़िकल यूज़र इंटरफ़ेस (GUI) होते हैं जहाँ आप बटन क्लिक करते हैं, पाठ इनपुट करते हैं, विकल्प चुनते हैं आदि।

### कंसोल और UI ऐप्स सीमित होते हैं

इसे एक कमांड-आधारित ऐप से तुलना करें जहां आप एक कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड टाइप नहीं कर सकते, केवल वे जो ऐप समर्थन करता है।
- **भाषा विशेष**। कुछ ऐप कई भाषाओं का समर्थन करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप किसी विशेष भाषा के लिए बनाया जाता है, भले ही आप अतिरिक्त भाषा समर्थन जोड़ सकें।

### पाठ उत्पन्न करने वाले ऐप के फायदे

तो एक पाठ उत्पन्न करने वाला ऐप कैसे अलग होता है?

एक पाठ उत्पन्न करने वाले ऐप में आपके पास अधिक लचीलापन होता है, आप कमांड सेट या किसी विशेष इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप प्राकृतिक भाषा का उपयोग करके ऐप के साथ इंटरैक्ट कर सकते हैं। एक अन्य लाभ यह है कि आप पहले से ही एक डाटा स्रोत के साथ इंटरैक्ट कर रहे हैं जिसे विशाल जानकारी के संग्रह पर प्रशिक्षित किया गया है, जबकि पारंपरिक ऐप डेटाबेस में सीमित हो सकता है।

### मैं पाठ उत्पन्न करने वाले ऐप से क्या बना सकता हूँ?

आप कई चीज़ें बना सकते हैं। उदाहरण के लिए:

- **एक चैटबोट**। एक चैटबोट जो विषयों जैसे आपकी कंपनी और इसके उत्पादों के बारे में प्रश्नों के उत्तर देता है, अच्छा विकल्प हो सकता है।
- **सहायक**। LLMs जैसे कार्यों में महान हैं, जैसे पाठ का सारांश बनाना, पाठ से सूचनाएँ निकालना, रिज्यूमे जैसे पाठ तैयार करना आदि।
- **कोड सहायक**। आपके द्वारा उपयोग किए जाने वाले भाषा मॉडल के आधार पर, आप ऐसा कोड सहायक बना सकते हैं जो आपको कोड लिखने में मदद करे। उदाहरण के लिए, आप GitHub Copilot या ChatGPT जैसे उत्पादों का उपयोग कर सकते हैं।

## मैं कैसे शुरू कर सकता हूँ?

खैर, आपको एक LLM के साथ इंटीग्रेट करने का तरीका ढूंढ़ना होगा, जिसमें आमतौर पर निम्न दो दृष्टिकोण शामिल होते हैं:

- API का उपयोग करें। यहां आप अपने प्रॉम्प्ट के साथ वेब अनुरोधों का निर्माण करते हैं और उत्पन्न पाठ वापस पाते हैं।
- एक लाइब्रेरी का उपयोग करें। लाइब्रेरी API कॉल्स को समेटने में मदद करती हैं और इन्हें उपयोग करना आसान बनाती हैं।

## लाइब्रेरियाँ/SDKs

LLM के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरी हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करने और प्रॉम्प्ट भेजने को आसान बनाती है।

फिर ऐसी लाइब्रेरियां भी हैं जो उच्च स्तरीय पर काम करती हैं जैसे:

- **Langchain**. Langchain प्रसिद्ध है और Python का समर्थन करता है।
- **Semantic Kernel**. Semantic Kernel माइक्रोसॉफ्ट की एक लाइब्रेरी है जो C#, Python और Java भाषाओं का समर्थन करती है।

## openai का उपयोग करके पहला ऐप

आइए देखें कि हम अपना पहला ऐप कैसे बना सकते हैं, हमें कौन-कौन सी लाइब्रेरी की आवश्यकता है, कितना आवश्यक है आदि।

### openai इंस्टॉल करें

  > [!NOTE] यदि आप यह नोटबुक Codespaces या Devcontainer में चला रहे हैं तो यह चरण आवश्यक नहीं है


OpenAI या Azure OpenAI के साथ इंटरैक्ट करने के लिए कई लाइब्रेरी उपलब्ध हैं। कई प्रोग्रामिंग भाषाओं जैसे C#, Python, JavaScript, Java और अधिक का उपयोग करना संभव है।  
हमने `openai` Python लाइब्रेरी का चयन किया है, इसलिए हम इसे इंस्टॉल करने के लिए `pip` का उपयोग करेंगे।

```bash
pip install openai
```

यदि आप यह नोटबुक Codespaces या Dev Container में नहीं चला रहे हैं, तो आपको अपनी मशीन पर [Python](https://www.python.org/) भी इंस्टॉल करना होगा।

### एक संसाधन बनाएं और अपनी API कुंजी ढूंढ़ें

यदि आपने पहले से नहीं किया है, तो आपको निम्नलिखित चरण करने होंगे:

- OpenAI पर एक खाता बनाएं <https://platform.openai.com/signup>।
- अब, अपनी API कुंजी प्राप्त करें <https://platform.openai.com/api-keys>। 

>[!NOTE]
> अपने कोड से अपनी API कुंजी अलग रखना बेहतर होता है। आप इसे पर्यावरण चर (environment variables) का उपयोग करके कर सकते हैं।
> - अपनी .env फ़ाइल में `OPENAI_KEY` पर्यावरण चर को अपनी API कुंजी के लिए सेट करें। यदि आपने इस पाठ्यक्रम के पिछले व्यायाम पूरे कर लिए हैं, तो आप पूरी तरह से सेट हैं।
> - यह ध्यान रखना महत्वपूर्ण है कि API कुंजी केवल एक बार उपलब्ध होगी। इसलिए, यह सुनिश्चित करना आवश्यक है कि इसे सही तरीके से कॉपी किया गया है। यदि यह अपेक्षित रूप से कार्य नहीं करता है, तो कुंजी को हटाना और एक नई जनरेट करना सुझाव दिया जाता है।


### OpenAI के लिए कॉन्फ़िगरेशन सेट करें

यदि आप OpenAI का उपयोग कर रहे हैं, तो इसे सेटअप करने का तरीका यह है:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

ऊपर हम निम्नलिखित सेट कर रहे हैं:

- `api_key`, यह आपकी OpenAI डैशबोर्ड में मिली API कुंजी है।
- `deployment`, यह आपका GPT संस्करण है।

> [!NOTE]
> `os.environ` एक फ़ंक्शन है जो पर्यावरण चर पढ़ता है। आप इसका उपयोग `OPENAI_API_KEY` जैसे पर्यावरण चर पढ़ने के लिए कर सकते हैं।

## टेक्स्ट जनरेट करें

टेक्स्ट उत्पन्न करने का तरीका है `responses.create` मेथड का उपयोग करना। यहाँ एक उदाहरण है:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ऊपर दिए गए कोड में, हम एक response ऑब्जेक्ट बनाते हैं और उस मॉडल और प्रॉम्प्ट को पास करते हैं जिसका उपयोग करना चाहते हैं। फिर हम उत्पन्न पाठ प्रिंट करते हैं।

### चैट प्रतिक्रियाएँ

अब तक, आपने देखा कि हम एक प्रॉम्प्ट से पाठ कैसे उत्पन्न कर सकते हैं। Responses API चैटबॉट्स के लिए भी उपयुक्त है - आप `input` के रूप में संदेशों की सूची प्रदान करते हैं ताकि बातचीत का निर्माण हो। इसका उपयोग करने का एक उदाहरण यहाँ है:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

इस फ़ंक्शनालिटी के बारे में अधिक जानकारी आने वाले अध्याय में मिलेगी।

## अभ्यास - आपका पहला पाठ उत्पन्न करने वाला ऐप

अब जब हमने जाना कि OpenAI सेवा को कैसे सेटअप और कॉन्फ़िगर किया जाता है, तो आपके लिए अपना पहला पाठ उत्पन्न करने वाला ऐप बनाने का समय है। अपने ऐप को बनाने के लिए, इन चरणों का पालन करें:


1. एक वर्चुअल वातावरण बनाएं और openai इंस्टॉल करें:

  > [!NOTE] यदि आप इस नोटबुक को Codespaces पर या Devcontainer के भीतर चला रहे हैं तो यह चरण आवश्यक नहीं है


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> यदि आप विंडोज़ का उपयोग कर रहे हैं तो `source venv/bin/activate` के बजाय `venv\Scripts\activate` टाइप करें। 

> [!NOTE]
> https://platform.openai.com/settings/organization/api-keys पर जाकर अपने OpenAI कुंजी का पता लगाएं और `API keys` खोजें। आप वहां एक नई कुंजी बना सकते हैं और तुरंत इसका मान कॉपी कर सकते हैं।


1. एक *app.py* फाइल बनाएं और उसमें निम्नलिखित कोड दें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    आपको निम्नलिखित जैसा आउटपुट दिखाई देना चाहिए:

    ```output
     बहुत दुखी _____.

    एक बार की बात है एक बहुत दुखी जलपरी थी।
    ```


## अलग-अलग चीजों के लिए अलग-अलग प्रकार के प्रॉम्प्ट

अब आपने देखा कि प्रॉम्प्ट का उपयोग करके कैसे टेक्स्ट जनरेट किया जाता है। आपके पास एक प्रोग्राम भी चल रहा है जिसे आप संशोधित और बदल सकते हैं ताकि विभिन्न प्रकार के टेक्स्ट बनाए जा सकें।

प्रॉम्प्ट्स का उपयोग सभी प्रकार के कार्यों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी प्रकार का टेक्स्ट जनरेट करें**। उदाहरण के लिए, आप एक कविता, क्विज़ के लिए प्रश्न आदि जनरेट कर सकते हैं।
- **जानकारी ढूंढना**। आप प्रॉम्प्ट का उपयोग जानकारी खोजने के लिए कर सकते हैं जैसे निम्न उदाहरण 'वेब विकास में CORS का क्या अर्थ है?'।
- **कोड जनरेट करें**। आप प्रॉम्प्ट्स का उपयोग कोड जनरेट करने के लिए कर सकते हैं, उदाहरण के लिए ईमेल वैलिडेशन के लिए एक रेगुलर एक्सप्रेशन विकसित करना या पूरा प्रोग्राम बनाना, जैसे एक वेब ऐप?

## एक अधिक व्यावहारिक उपयोग मामला: एक रेसिपी जनरेटर

कल्पना करें कि आपके पास घर पर सामग्री है और आप कुछ पकाना चाहते हैं। इसके लिए, आपको एक रेसिपी की जरूरत है। रेसिपी खोजने का एक तरीका सर्च इंजन का उपयोग करना है या आप इसके लिए एक LLM का उपयोग कर सकते हैं।

आप ऐसा प्रॉम्प्ट लिख सकते हैं:

> "मुझे निम्नलिखित सामग्री के साथ एक व्यंजन के लिए 5 रेसिपी दिखाएं: चिकन, आलू, और गाजर। प्रति रेसिपी, उपयोग की गई सभी सामग्री सूचीबद्ध करें"

ऊपर दिए गए प्रॉम्प्ट को देखते हुए, आपको संभवतः इस तरह का जवाब मिल सकता है:

```output
1. Roasted Chicken and Vegetables: 
Ingredients: 
- 4 chicken thighs
- 2 potatoes, cut into cubes
- 2 carrots, cut into cubes
- 2 tablespoons olive oil
- 2 cloves garlic, minced
- 1 teaspoon dried thyme
```

## अभ्यास - एक रेसिपी जनरेटर बनाएं

अब जब हमने एक परिदृश्य को खेल लिया है, तो चलिए उस प्रदर्शित परिदृश्य से मेल खाने वाला कोड लिखते हैं। ऐसा करने के लिए, निम्नलिखित चरणों का पालन करें:

1. शुरुआत के लिए मौजूदा *app.py* फ़ाइल का उपयोग करें
1. `prompt` वेरिएबल को ढूंढें और उसके कोड को निम्नानुसार बदलें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

यदि आप अब कोड चलाते हैं, तो आपको कुछ इसी तरह का आउटपुट देखने को मिलेगा:

> नोट करें, आपका LLM गैर-नियतात्मक है, इसलिए आप प्रोग्राम को हर बार चलाने पर अलग-अलग परिणाम प्राप्त कर सकते हैं।

बढ़िया, आइए देखें कि हम चीज़ों को कैसे बेहतर बना सकते हैं। चीज़ों को बेहतर बनाने के लिए, हम यह सुनिश्चित करना चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या को बेहतर और बदला जा सके। 


1. आइए कोड को निम्नलिखित तरीके से बदलें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### फ़िल्टर और शॉपिंग लिस्ट जोड़कर सुधार करें

हमारे पास अब एक कार्यशील ऐप है जो रेसिपी बना सकता है और यह लचीला है क्योंकि यह उपयोगकर्ता से इनपुट पर निर्भर करता है, न केवल रेसिपी की संख्या पर बल्कि उपयोग की गई सामग्री पर भी।

इसे और बेहतर बनाने के लिए, हम निम्नलिखित जोड़ना चाहते हैं:

- **सामग्री को फ़िल्टर करें**। हम उन सामग्री को फ़िल्टर करना चाहते हैं जिन्हें हम पसंद नहीं करते या जिनसे हमें एलर्जी है। इस बदलाव को पूरा करने के लिए, हम अपने मौजूदा प्रॉम्प्ट को संपादित कर सकते हैं और इसके अंत में एक फ़िल्टर शर्त जोड़ सकते हैं, इस प्रकार:

    ```python

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    ऊपर, हमने `{filter}` को प्रॉम्प्ट के अंत में जोड़ा है और हम उपयोगकर्ता से फ़िल्टर मान भी प्राप्त करते हैं।

    जैसा कि आप देख सकते हैं, दूध वाली कोई भी रेसिपी फ़िल्टर कर दी गई है। लेकिन, अगर आप लैक्टोज़ असहिष्णु हैं, तो आप पनीर वाली रेसिपी को भी फिल्टर करना चाहेंगे, इसलिए यहां स्पष्ट होने की आवश्यकता है।

- **एक शॉपिंग लिस्ट बनाएं**। हम एक शॉपिंग लिस्ट बनाना चाहते हैं, यह ध्यान में रखते हुए कि हमारे पास घर पर क्या पहले से है।

    इस कार्यक्षमता के लिए, हम या तो सब कुछ एक प्रॉम्प्ट में हल करने की कोशिश कर सकते हैं या इसे दो प्रॉम्प्ट में विभाजित कर सकते हैं। चलिए बाद वाला तरीका आजमाते हैं। यहां हम एक अतिरिक्त प्रॉम्प्ट जोड़ने का सुझाव दे रहे हैं, लेकिन इसके काम करने के लिए, हमें पहले प्रॉम्प्ट के परिणाम को बाद वाले प्रॉम्प्ट के संदर्भ के रूप में जोड़ना होगा।

    वह हिस्सा ढूंढें जो कोड में पहला प्रॉम्प्ट का परिणाम प्रिंट करता है और उसके नीचे निम्न कोड जोड़ें:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # प्रतिक्रिया प्रिंट करें
    print("Shopping list:")
    print(response.output_text)
    ```

    निम्न बातों का ध्यान रखें:

    - हम एक नया प्रॉम्प्ट बना रहे हैं जिसमें पहले प्रॉम्प्ट का परिणाम नया प्रॉम्प्ट है:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - हम एक नया अनुरोध करते हैं, लेकिन पहले प्रॉम्प्ट में हमने जो टोकन मांगे थे उसे ध्यान में रखते हुए, इस बार हम कहते हैं `max_output_tokens` 1200 है।

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        इस कोड को चलाने पर हमें अब निम्नलिखित आउटपुट मिलता है:


- **टोकन की लंबाई पर एक शब्द**। हमें यह विचार करना चाहिए कि हमें कितना टोकन चाहिए उस टेक्स्ट को जनरेट करने के लिए जो हम चाहते हैं। टोकन की लागत होती है, इसलिए जहां संभव हो, हमें टोकन की संख्या में आर्थिक होना चाहिए। उदाहरण के लिए, क्या हम प्रॉम्प्ट को इस तरह बना सकते हैं कि कम टोकन इस्तेमाल हों?

   टोकन की संख्या बदलने के लिए, आप `max_output_tokens` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, अगर आप 100 टोकन उपयोग करना चाहते हैं, तो आप करेंगे:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **थर्मल (temperature) के साथ प्रयोग**। थर्मल एक ऐसा पैरामीटर है जिसके बारे में हमने अब तक बात नहीं की है लेकिन यह हमारे प्रोग्राम के प्रदर्शन के लिए महत्वपूर्ण है। थर्मल का मूल्य जितना बड़ा होगा, आउटपुट उतना ही अधिक यादृच्छिक होगा। जबकि थर्मल का मूल्य जितना कम होगा, आउटपुट उतना ही अधिक पूर्वानुमानित होगा। सोचें कि क्या आप अपने आउटपुट में विविधता चाहते हैं या नहीं।

   थर्मल को बदलने के लिए, आप `temperature` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, अगर आप 0.5 का तापमान चाहते हैं, तो आप करेंगे:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > नोट करें, 1.0 के जितना करीब होगा, आउटपुट उतना ही अधिक विविध होगा।
