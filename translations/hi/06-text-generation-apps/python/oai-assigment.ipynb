# टेक्स्ट जनरेशन ऐप्स बनाएं

अब तक आपने इस पाठ्यक्रम में देखा है कि कुछ मुख्य कॉन्सेप्ट्स होते हैं जैसे कि प्रॉम्प्ट्स और एक पूरी डिसिप्लिन है जिसे "प्रॉम्प्ट इंजीनियरिंग" कहा जाता है। कई टूल्स जैसे ChatGPT, Office 365, Microsoft Power Platform आदि, आपको प्रॉम्प्ट्स का इस्तेमाल करके कुछ हासिल करने में मदद करते हैं।

अगर आप अपने ऐप में ऐसा अनुभव जोड़ना चाहते हैं, तो आपको प्रॉम्प्ट्स, कम्प्लीशन जैसे कॉन्सेप्ट्स समझने होंगे और काम करने के लिए एक लाइब्रेरी चुननी होगी। इसी बारे में आप इस चैप्टर में सीखेंगे।

## परिचय

इस चैप्टर में, आप सीखेंगे:

- openai लाइब्रेरी और उसके मुख्य कॉन्सेप्ट्स के बारे में।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाना।
- प्रॉम्प्ट, टेम्परेचर और टोकन्स जैसे कॉन्सेप्ट्स का इस्तेमाल करके टेक्स्ट जनरेशन ऐप कैसे बनाएं, यह समझना।

## लर्निंग गोल्स

इस पाठ के अंत में, आप सक्षम होंगे:

- समझा पाएंगे कि टेक्स्ट जनरेशन ऐप क्या है।
- openai का उपयोग करके टेक्स्ट जनरेशन ऐप बना पाएंगे।
- अपने ऐप को ज्यादा या कम टोकन्स इस्तेमाल करने के लिए और आउटपुट में विविधता लाने के लिए टेम्परेचर बदलने के लिए कॉन्फ़िगर कर पाएंगे।

## टेक्स्ट जनरेशन ऐप क्या है?

आमतौर पर जब आप कोई ऐप बनाते हैं तो उसमें कुछ इस तरह का इंटरफेस होता है:

- कमांड-बेस्ड। कंसोल ऐप्स आमतौर पर ऐसे ऐप्स होते हैं जहां आप कोई कमांड टाइप करते हैं और वह कोई टास्क करता है। उदाहरण के लिए, `git` एक कमांड-बेस्ड ऐप है।
- यूजर इंटरफेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUI) होता है जिसमें आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, ऑप्शन चुनते हैं आदि।

### कंसोल और UI ऐप्स की सीमाएं

इसे एक कमांड-बेस्ड ऐप से तुलना करें जिसमें आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड नहीं टाइप कर सकते, सिर्फ वही जो ऐप सपोर्ट करता है।
- **भाषा-विशिष्ट**। कुछ ऐप्स कई भाषाओं को सपोर्ट करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप एक खास भाषा के लिए बना होता है, भले ही आप और भाषाओं का सपोर्ट जोड़ सकते हैं।

### टेक्स्ट जनरेशन ऐप्स के फायदे

तो टेक्स्ट जनरेशन ऐप किस तरह अलग है?

टेक्स्ट जनरेशन ऐप में आपके पास ज्यादा लचीलापन होता है, आप सिर्फ कुछ कमांड्स या किसी खास इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप नेचुरल लैंग्वेज का इस्तेमाल करके ऐप से इंटरैक्ट कर सकते हैं। एक और फायदा यह है कि आप ऐसे डेटा सोर्स से इंटरैक्ट कर रहे हैं जिसे बहुत बड़ी जानकारी पर ट्रेन किया गया है, जबकि पारंपरिक ऐप्स सिर्फ डेटाबेस में मौजूद जानकारी तक सीमित हो सकते हैं।

### टेक्स्ट जनरेशन ऐप से क्या-क्या बना सकते हैं?

आप बहुत कुछ बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। कोई चैटबॉट जो आपकी कंपनी और उसके प्रोडक्ट्स जैसे टॉपिक्स पर सवालों के जवाब दे सकता है, एक अच्छा विकल्प हो सकता है।
- **हेल्पर**। LLMs टेक्स्ट को संक्षिप्त करने, टेक्स्ट से इनसाइट्स निकालने, रिज्यूमे जैसे टेक्स्ट बनाने आदि में बहुत अच्छे हैं।
- **कोड असिस्टेंट**। आप जिस लैंग्वेज मॉडल का इस्तेमाल करते हैं, उसके आधार पर आप एक कोड असिस्टेंट बना सकते हैं जो कोड लिखने में मदद करे। उदाहरण के लिए, आप GitHub Copilot या ChatGPT जैसे प्रोडक्ट्स का इस्तेमाल कोड लिखने में मदद के लिए कर सकते हैं।

## कैसे शुरू करें?

आपको LLM के साथ इंटीग्रेट करने का तरीका ढूंढना होगा, जो आमतौर पर दो तरीकों से होता है:

- API का इस्तेमाल करें। इसमें आप अपनी प्रॉम्प्ट के साथ वेब रिक्वेस्ट बनाते हैं और जनरेटेड टेक्स्ट वापस पाते हैं।
- लाइब्रेरी का इस्तेमाल करें। लाइब्रेरीज़ API कॉल्स को आसान बनाती हैं और इस्तेमाल में सरल बनाती हैं।

## लाइब्रेरी/SDKs

LLMs के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरीज़ हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करना और प्रॉम्प्ट भेजना आसान बनाती है।

फिर कुछ लाइब्रेरीज़ हैं जो और ऊँचे स्तर पर काम करती हैं जैसे:

- **Langchain**। Langchain काफी प्रसिद्ध है और Python को सपोर्ट करता है।
- **Semantic Kernel**। Semantic Kernel माइक्रोसॉफ्ट की लाइब्रेरी है जो C#, Python और Java को सपोर्ट करती है।

## openai का उपयोग करके पहला ऐप

आइए देखें कि हम अपना पहला ऐप कैसे बना सकते हैं, हमें कौन सी लाइब्रेरी चाहिए, कितना कोड लिखना होगा आदि।

### openai इंस्टॉल करें

  > [!NOTE] यह स्टेप जरूरी नहीं है अगर आप यह नोटबुक Codespaces या Devcontainer में चला रहे हैं


OpenAI या Azure OpenAI के साथ इंटरैक्ट करने के लिए कई लाइब्रेरीज़ उपलब्ध हैं। आप C#, Python, JavaScript, Java आदि जैसी कई प्रोग्रामिंग भाषाओं का भी इस्तेमाल कर सकते हैं।  
हमने `openai` Python लाइब्रेरी चुनी है, इसलिए हम इसे इंस्टॉल करने के लिए `pip` का इस्तेमाल करेंगे।

### रिसोर्स बनाएं और अपना API की ढूंढें

अगर आपने पहले से नहीं किया है, तो आपको ये स्टेप्स करने होंगे:

- OpenAI पर अकाउंट बनाएं <https://platform.openai.com/signup>।
- अब, अपनी API कीज प्राप्त करें <https://platform.openai.com/api-keys>। 

>[!NOTE]
> अपने API की कोड से अलग रखना अच्छा रहता है। आप ऐसा environment variables का इस्तेमाल करके कर सकते हैं।
> - अपनी .env फाइल में environment variable `OPENAI_KEY` को अपनी API की से सेट करें। अगर आपने इस कोर्स के पिछले एक्सरसाइज पूरे कर लिए हैं, तो आप तैयार हैं।
> - ध्यान देने वाली बात है कि API की सिर्फ एक बार ही एक्सेस की जा सकती है। इसलिए यह सुनिश्चित करना जरूरी है कि आपने उसे सही से कॉपी किया है। अगर वह काम नहीं करती, तो की को डिलीट करके नई की जनरेट करें।


### OpenAI की कॉन्फ़िगरेशन सेटअप करें

अगर आप OpenAI का इस्तेमाल कर रहे हैं, तो इसकी कॉन्फ़िगरेशन ऐसे सेट करें:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

ऊपर हम ये सेट कर रहे हैं:

- `api_key`, यह आपकी API की है जो OpenAI डैशबोर्ड में मिलती है।
- `deployment`, यह आपके GPT वर्शन का नाम है।

> [!NOTE]
> `os.environ` एक फंक्शन है जो environment variables पढ़ता है। आप इसका इस्तेमाल `OPENAI_API_KEY` जैसे environment variables पढ़ने के लिए कर सकते हैं।

## टेक्स्ट जनरेट करें

टेक्स्ट जनरेट करने का तरीका है `chat.completion` क्लास का इस्तेमाल करना। उदाहरण देखें:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

ऊपर दिए गए कोड में, हम एक कम्प्लीशन ऑब्जेक्ट बनाते हैं और उसमें मॉडल और प्रॉम्प्ट पास करते हैं। फिर हम जनरेटेड टेक्स्ट प्रिंट करते हैं।

### चैट कम्प्लीशन

अब तक आपने देखा कि हम `Completion` का इस्तेमाल टेक्स्ट जनरेट करने के लिए कर रहे थे। लेकिन एक और क्लास है `ChatCompletion` जो चैटबॉट्स के लिए ज्यादा उपयुक्त है। इसका उदाहरण देखें:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

इस फंक्शनैलिटी के बारे में आगे के चैप्टर में और जानेंगे।

## अभ्यास - अपना पहला टेक्स्ट जनरेशन ऐप

अब जब हमने OpenAI सर्विस को सेटअप और कॉन्फ़िगर करना सीख लिया है, तो अब बारी है अपना पहला टेक्स्ट जनरेशन ऐप बनाने की। अपना ऐप बनाने के लिए ये स्टेप्स फॉलो करें:


1. एक वर्चुअल एनवायरनमेंट बनाएं और openai इंस्टॉल करें:

  > [!NOTE] यह स्टेप जरूरी नहीं है अगर आप यह नोटबुक Codespaces या Devcontainer में चला रहे हैं


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> अगर आप Windows इस्तेमाल कर रहे हैं तो `source venv/bin/activate` की जगह `venv\Scripts\activate` टाइप करें।

> [!NOTE]
> अपना OpenAI key खोजने के लिए https://platform.openai.com/settings/organization/api-keys पर जाएँ और `API keys` सर्च करें। आप वहाँ नया key बना सकते हैं और तुरंत उसकी वैल्यू कॉपी कर सकते हैं।


1. एक *app.py* फ़ाइल बनाएं और उसमें निम्नलिखित कोड दें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## अलग-अलग चीज़ों के लिए अलग-अलग प्रकार के प्रॉम्प्ट्स

अब आपने देखा कि कैसे एक प्रॉम्प्ट का इस्तेमाल करके टेक्स्ट जनरेट किया जा सकता है। आपके पास एक प्रोग्राम भी है जिसे आप बदल सकते हैं और अलग-अलग तरह का टेक्स्ट बना सकते हैं।

प्रॉम्प्ट्स का इस्तेमाल कई तरह के कामों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी खास तरह का टेक्स्ट जनरेट करना**। जैसे, आप एक कविता बना सकते हैं, क्विज़ के लिए सवाल तैयार कर सकते हैं आदि।
- **जानकारी ढूंढना**। आप प्रॉम्प्ट्स का इस्तेमाल जानकारी खोजने के लिए कर सकते हैं, जैसे 'वेब डेवलपमेंट में CORS का क्या मतलब है?'।
- **कोड जनरेट करना**। आप प्रॉम्प्ट्स की मदद से कोड बना सकते हैं, जैसे ईमेल वैलिडेट करने के लिए रेगुलर एक्सप्रेशन तैयार करना या फिर पूरा प्रोग्राम जनरेट करना, जैसे एक वेब ऐप।

## अभ्यास - एक रेसिपी जनरेटर बनाएं

अब जब हमने एक परिदृश्य को देखा है, तो चलिए उस परिदृश्य के अनुसार कोड लिखते हैं। ऐसा करने के लिए, निम्नलिखित चरणों का पालन करें:

1. मौजूदा *app.py* फ़ाइल को शुरुआती बिंदु के रूप में उपयोग करें
1. `prompt` वेरिएबल को ढूंढें और उसके कोड को निम्नलिखित में बदलें:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

बहुत बढ़िया, चलिए देखते हैं कि हम चीजों को कैसे बेहतर बना सकते हैं। चीजों को बेहतर बनाने के लिए, हम चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या को आसानी से बदला और सुधारा जा सके।


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

    ```python
    
- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt.

    old_prompt_result = completion.choices[0].text

        completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.
