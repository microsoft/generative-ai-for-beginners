# टेक्स्ट जेनरेशन ऐप बनाएं

आपने अब तक इस पाठ्यक्रम के माध्यम से देखा होगा कि कोर कॉन्सेप्ट्स जैसे प्रॉम्प्ट्स और यहाँ तक कि "prompt engineering" नामक एक पूरा विषय है। कई टूल्स जिनसे आप इंटरैक्ट कर सकते हैं जैसे ChatGPT, Office 365, Microsoft Power Platform और अधिक, आपको कुछ पूरा करने के लिए प्रॉम्प्ट्स का उपयोग करने का समर्थन करते हैं।

आपकी ऐप में ऐसा अनुभव जोड़ने के लिए, आपको प्रॉम्प्ट्स, कंप्लीशंस जैसी अवधारणाओं को समझना होगा और काम करने के लिए एक लाइब्रेरी चुननी होगी। यही आप इस अध्याय में सीखेंगे।

## परिचय

इस अध्याय में, आप:

- openai लाइब्रेरी और इसके मुख्य कॉन्सेप्ट्स के बारे में सीखेंगे।
- openai का उपयोग करके एक टेक्स्ट जेनरेशन ऐप बनाएंगे।
- प्रॉम्प्ट, टेम्परेचर, और टोकन जैसे कॉन्सेप्ट्स का उपयोग करके टेक्स्ट जेनरेशन ऐप बनाने का तरीका समझेंगे।

## सीखने के लक्ष्य

इस पाठ के अंत में, आप सक्षम होंगे:

- टेक्स्ट जेनरेशन ऐप क्या होता है, समझाएं।
- openai का उपयोग करके एक टेक्स्ट जेनरेशन ऐप बनाएं।
- अपने ऐप को अधिक या कम टोकन उपयोग करने और टेम्परेचर बदलने के लिए कॉन्फ़िगर करें, ताकि परिणाम में विविधता हो।

## टेक्स्ट जेनरेशन ऐप क्या है?

सामान्य रूप से जब आप कोई ऐप बनाते हैं, तो उसमे कुछ प्रकार का इंटरफ़ेस होता है जैसे नीचे दिया गया:

- कमांड-आधारित। कंसोल ऐप्स मूलतः ऐसे ऐप होते हैं जहाँ आप कमांड टाइप करते हैं और यह कोई कार्य करता है। उदाहरण के लिए, `git` एक कमांड-आधारित ऐप है।
- यूजर इंटरफ़ेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUIs) होते हैं जहाँ आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं आदि।

### कंसोल और UI ऐप्स सीमित हैं

इसे कमांड-आधारित ऐप से तुलना करें जहाँ आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड टाइप नहीं कर सकते, केवल वही जो ऐप सपोर्ट करता है।
- **भाषा विशिष्ट**। कुछ ऐप्स कई भाषाओं का समर्थन करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप किसी विशिष्ट भाषा के लिए बनाया गया होता है, भले ही आप और भाषाओं को जोड़ सकें।

### टेक्स्ट जेनरेशन ऐप के लाभ

तो टेक्स्ट जेनरेशन ऐप कैसे अलग है?

टेक्स्ट जेनरेशन ऐप में आपको अधिक लचीलापन मिलता है, आप कमांड सेट या किसी विशेष इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप प्राकृतिक भाषा का उपयोग करके ऐप से इंटरैक्ट कर सकते हैं। एक अन्य लाभ यह है कि आप उस डेटा स्रोत के साथ पहले से ही बातचीत कर रहे हैं जिसे विशाल जानकारी के कार्पस पर प्रशिक्षित किया गया है, जबकि पारंपरिक ऐप शायद केवल उस डाटाबेस तक सीमित होते हैं जो उसके पास है।

### मैं टेक्स्ट जेनरेशन ऐप से क्या बना सकता हूँ?

आप कई प्रकार की चीजें बना सकते हैं। उदाहरण के लिए:

- **एक चैटबोट**। आपकी कंपनी और उसके उत्पादों के विषय में सवालों के जवाब देने वाला चैटबोट एक अच्छा विकल्प हो सकता है।
- **हेल्पर**। LLMs जैसे बड़े भाषा मॉडल टेक्स्ट का सारांश बनाने, टेक्स्ट से जानकारी निकालने, रिज्यूमे जैसी टेक्स्ट सामग्री उत्पन्न करने में बहुत अच्छे हैं।
- **कोड असिस्टेंट**। जिस भाषा मॉडल का आप उपयोग करते हैं, उसके आधार पर आप एक कोड असिस्टेंट बना सकते हैं जो कोड लिखने में आपकी मदद करता है। उदाहरण के लिए, आप GitHub Copilot और ChatGPT जैसे उत्पादों का उपयोग कर सकते हैं।

## मैं कैसे शुरू कर सकता हूँ?

आपको किसी LLM के साथ इंटीग्रेट करने का तरीका ढूंढ़ना होगा, जो आमतौर पर निम्न दो तरीकों से होता है:

- API का उपयोग करें। यहाँ आप अपने प्रॉम्प्ट के साथ वेब रिक्वेस्ट बनाते हैं और जेनरेट किया गया टेक्स्ट प्राप्त करते हैं।
- लाइब्रेरी का उपयोग करें। लाइब्रेरी API कॉल्स को कैप्सुलेट करती हैं और इन्हें उपयोग में आसान बनाती हैं।

## लाइब्रेरी/SDKs

LLMs के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरी हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से जुड़ने और प्रॉम्प्ट भेजने में आसानी प्रदान करती है।

और कुछ ऐसी लाइब्रेरी हैं जो उच्च स्तर पर काम करती हैं जैसे:

- **Langchain**। Langchain प्रसिद्ध है और Python को सपोर्ट करता है।
- **Semantic Kernel**। Semantic Kernel माइक्रोसॉफ्ट की लाइब्रेरी है जो C#, Python, और Java भाषाओं का समर्थन करती है।

## ChatCompletions के साथ टेक्स्ट जेनरेट करें

टेक्स्ट जेनरेट करने का तरीका `ChatCompletionsClient` क्लास का उपयोग करना है।
`samples/python/azure_ai_inference/basic.py` में, कोड के response सेक्शन में, user role के कोड को अपडेट करें और content पैरामीटर को नीचे दिए अनुसार बदलें:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # वैकल्पिक पैरामीटर
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

आउटपुट देखने के लिए अपडेटेड फाइल चलाएं


## विभिन्न चीज़ों के लिए विभिन्न प्रकार के प्रॉम्प्ट

अब आपने देखा है कि प्रॉम्प्ट का उपयोग करके टेक्स्ट कैसे उत्पन्न किया जाता है। आपके पास एक प्रोग्राम भी है जो चल रहा है जिसे आप अलग-अलग प्रकार का टेक्स्ट उत्पन्न करने के लिए संशोधित और बदल सकते हैं। 

प्रॉम्प्ट का उपयोग कई प्रकार के कार्यों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी प्रकार का टेक्स्ट उत्पन्न करें**। उदाहरण के लिए, आप एक कविता, क्विज़ के सवाल आदि उत्पन्न कर सकते हैं।
- **जानकारी खोजें**। आप प्रॉम्प्ट का उपयोग सूचना खोजने के लिए कर सकते हैं, जैसे निम्न उदाहरण 'वेब विकास में CORS का क्या अर्थ है?'।
- **कोड उत्पन्न करें**। आप प्रॉम्प्ट का उपयोग कोड उत्पन्न करने के लिए कर सकते हैं, उदाहरण के लिए ईमेल की जांच के लिए रेगुलर एक्सप्रेशन विकसित करना या एक पूरा प्रोग्राम, जैसे एक वेब ऐप, उत्पन्न करना।

## अभ्यास - एक रेसिपी जनरेटर बनाएँ

अब जब हमने एक स्थिति का प्रदर्शन कर दिया है, तो चलिए उस प्रदर्शित स्थिति से मेल खाने वाला कोड लिखते हैं। ऐसा करने के लिए, निम्न चरणों का पालन करें:

1. मौजूदा फ़ाइल को प्रारंभिक बिंदु के रूप में उपयोग करें
1. एक `prompt` वेरिएबल बनाएं और उदाहरण कोड को नीचे दिए अनुसार बदलें:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

> नोट, आपका LLM गैर-नियतात्मक है, इसलिए आप हर बार प्रोग्राम चलाने पर अलग परिणाम प्राप्त कर सकते हैं।

बहुत बढ़िया, आइए देखें कि हम चीजों में कैसे सुधार कर सकते हैं। सुधार करने के लिए, हम यह सुनिश्चित करना चाहते हैं कि कोड लचीला हो, ताकि सामग्री और नुस्खों की संख्या में सुधार और बदलाव किया जा सके। 


1. आइए कोड को निम्नलिखित तरीके से बदलते हैं:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **तापमान के साथ प्रयोग करना**। तापमान एक ऐसी चीज है जिसका हमने अब तक उल्लेख नहीं किया है लेकिन यह महत्वपूर्ण संदर्भ है कि हमारा प्रोग्राम कैसे प्रदर्शन करता है। तापमान मान जितना अधिक होगा, आउटपुट उतना ही यादृच्छिक होगा। इसके विपरीत, तापमान मान जितना कम होगा, आउटपुट उतना ही पूर्वानुमेय होगा। विचार करें कि क्या आप अपने आउटपुट में विविधता चाहते हैं या नहीं।

   तापमान परिवर्तित करने के लिए, आप `temperature` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, यदि आप 0.5 तापमान का उपयोग करना चाहते हैं, तो आप ऐसा करेंगे:

```python
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
            },
            {
                "role": "user",
                "content": new_prompt,
            },
        ],
        model=model_name,
        # वैकल्पिक पैरामीटर
        temperature=0.5,
        max_tokens=1200,
        top_p=1.    
    )
```

   > नोट करें, 1.0 के जितना निकट होगा, आउटपुट उतना ही विविध होगा।

- एक "स्टडी बडी" बनाएं। यह ऐप किसी विषय, उदाहरण के लिए Python के बारे में प्रश्नों के उत्तर देने में सक्षम होना चाहिए, आपके पास ऐसे प्रांप्ट हो सकते हैं जैसे "Python में कोई विशिष्ट विषय क्या है?", या कोई ऐसा प्रांप्ट जो कहता है, मुझे किसी विषय के लिए कोड दिखाओ आदि।
- इतिहास बॉट, इतिहास को जीवंत बनाएं, बॉट को निर्देश दें कि वह किसी विशिष्ट ऐतिहासिक व्यक्ति के रूप में कार्य करे और उससे उसके जीवन तथा समय के बारे में प्रश्न पूछें।
