# टेक्स्ट जनरेशन ऐप्स बनाएं

अब तक आपने इस पाठ्यक्रम में देखा है कि प्रॉम्प्ट जैसे कोर कॉन्सेप्ट्स होते हैं और यहां तक कि एक पूरी डिसिप्लिन है जिसे "प्रॉम्प्ट इंजीनियरिंग" कहा जाता है। कई टूल्स जैसे ChatGPT, Office 365, Microsoft Power Platform आदि, आपको प्रॉम्प्ट्स का इस्तेमाल करके कुछ हासिल करने में मदद करते हैं।

अगर आप अपने ऐप में ऐसा अनुभव जोड़ना चाहते हैं, तो आपको प्रॉम्प्ट्स, कम्प्लीशन जैसे कॉन्सेप्ट्स समझने होंगे और काम करने के लिए एक लाइब्रेरी चुननी होगी। यही आप इस चैप्टर में सीखेंगे।

## परिचय

इस चैप्टर में, आप:

- openai लाइब्रेरी और उसके मुख्य कॉन्सेप्ट्स के बारे में जानेंगे।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएंगे।
- प्रॉम्प्ट, टेम्परेचर और टोकन्स जैसे कॉन्सेप्ट्स का इस्तेमाल करके टेक्स्ट जनरेशन ऐप बनाना समझेंगे।

## लर्निंग गोल्स

इस पाठ के अंत में, आप सक्षम होंगे:

- समझा पाएंगे कि टेक्स्ट जनरेशन ऐप क्या है।
- openai का उपयोग करके टेक्स्ट जनरेशन ऐप बना पाएंगे।
- अपने ऐप को ज्यादा या कम टोकन्स इस्तेमाल करने के लिए और आउटपुट में विविधता लाने के लिए टेम्परेचर बदलने के लिए कॉन्फ़िगर कर पाएंगे।

## टेक्स्ट जनरेशन ऐप क्या है?

आमतौर पर जब आप कोई ऐप बनाते हैं तो उसमें कुछ तरह का इंटरफेस होता है, जैसे:

- कमांड-बेस्ड। कंसोल ऐप्स आमतौर पर वे ऐप्स होती हैं जहां आप एक कमांड टाइप करते हैं और वह कोई टास्क पूरा करती है। उदाहरण के लिए, `git` एक कमांड-बेस्ड ऐप है।
- यूजर इंटरफेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUI) होता है, जिसमें आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं आदि।

### कंसोल और UI ऐप्स की सीमाएं

इसे एक कमांड-बेस्ड ऐप से तुलना करें जिसमें आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड नहीं टाइप कर सकते, केवल वही जो ऐप सपोर्ट करता है।
- **भाषा-विशिष्ट**। कुछ ऐप्स कई भाषाओं को सपोर्ट करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप एक खास भाषा के लिए बना होता है, भले ही आप और भाषाओं का सपोर्ट जोड़ सकते हैं।

### टेक्स्ट जनरेशन ऐप्स के फायदे

तो टेक्स्ट जनरेशन ऐप किस तरह अलग है?

टेक्स्ट जनरेशन ऐप में आपके पास ज्यादा लचीलापन होता है, आप केवल कुछ कमांड्स या किसी खास इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप नेचुरल लैंग्वेज का इस्तेमाल करके ऐप से इंटरैक्ट कर सकते हैं। एक और फायदा यह है कि आप ऐसे डेटा सोर्स से इंटरैक्ट कर रहे हैं जिसे बहुत बड़े डाटा पर ट्रेन किया गया है, जबकि पारंपरिक ऐप्स केवल डेटाबेस में मौजूद जानकारी तक सीमित हो सकते हैं।

### टेक्स्ट जनरेशन ऐप से क्या-क्या बना सकते हैं?

आप बहुत कुछ बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। कोई चैटबॉट जो आपके कंपनी और उसके प्रोडक्ट्स के बारे में सवालों के जवाब दे सकता है, एक अच्छा उदाहरण हो सकता है।
- **हेल्पर**। LLMs टेक्स्ट को संक्षिप्त करने, टेक्स्ट से इनसाइट्स निकालने, रिज्यूमे जैसे टेक्स्ट बनाने आदि में बहुत अच्छे हैं।
- **कोड असिस्टेंट**। आप जिस लैंग्वेज मॉडल का इस्तेमाल करते हैं, उसके आधार पर आप एक कोड असिस्टेंट बना सकते हैं जो कोड लिखने में मदद करे। उदाहरण के लिए, आप GitHub Copilot या ChatGPT जैसे प्रोडक्ट्स का इस्तेमाल कोड लिखने में मदद के लिए कर सकते हैं।

## मैं कैसे शुरू करूं?

आपको LLM के साथ इंटीग्रेट करने का तरीका ढूंढना होगा, जो आमतौर पर दो तरीकों में से एक होता है:

- API का इस्तेमाल करें। इसमें आप अपनी प्रॉम्प्ट के साथ वेब रिक्वेस्ट बनाते हैं और जनरेटेड टेक्स्ट वापस पाते हैं।
- लाइब्रेरी का इस्तेमाल करें। लाइब्रेरीज़ API कॉल्स को आसान बनाती हैं और इस्तेमाल में सरल बनाती हैं।

## लाइब्रेरी/SDKs

LLMs के साथ काम करने के लिए कुछ प्रसिद्ध लाइब्रेरीज़ हैं, जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करना और प्रॉम्प्ट भेजना आसान बनाती है।

इसके अलावा कुछ लाइब्रेरीज़ हैं जो और ऊँचे स्तर पर काम करती हैं, जैसे:

- **Langchain**। Langchain काफी प्रसिद्ध है और Python को सपोर्ट करता है।
- **Semantic Kernel**। Semantic Kernel माइक्रोसॉफ्ट की लाइब्रेरी है जो C#, Python और Java को सपोर्ट करती है।

## ChatCompletions के साथ टेक्स्ट जनरेट करें

टेक्स्ट जनरेट करने का तरीका है `ChatCompletionsClient` क्लास का उपयोग करना।
`samples/python/azure_ai_inference/basic.py` में, कोड के response सेक्शन में, यूजर रोल के कंटेंट पैरामीटर को नीचे दिए गए अनुसार अपडेट करें:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट की गई फाइल को चलाएं और आउटपुट देखें


## अलग-अलग चीज़ों के लिए अलग-अलग प्रकार के प्रॉम्प्ट्स

अब आपने देखा कि कैसे एक प्रॉम्प्ट का इस्तेमाल करके टेक्स्ट जनरेट किया जा सकता है। आपके पास एक प्रोग्राम भी है जिसे आप बदल सकते हैं और अलग-अलग तरह का टेक्स्ट बना सकते हैं।

प्रॉम्प्ट्स का इस्तेमाल कई तरह के कामों के लिए किया जा सकता है। उदाहरण के लिए:

- **किसी खास तरह का टेक्स्ट जनरेट करना**। जैसे, आप एक कविता बना सकते हैं, क्विज़ के लिए सवाल तैयार कर सकते हैं आदि।
- **जानकारी ढूंढना**। आप प्रॉम्प्ट्स का इस्तेमाल जानकारी खोजने के लिए कर सकते हैं, जैसे 'वेब डेवलपमेंट में CORS का क्या मतलब है?'।
- **कोड जनरेट करना**। आप प्रॉम्प्ट्स की मदद से कोड बना सकते हैं, जैसे ईमेल वैलिडेट करने के लिए रेगुलर एक्सप्रेशन तैयार करना या फिर पूरा प्रोग्राम जनरेट करना, जैसे एक वेब ऐप।

## अभ्यास - एक रेसिपी जनरेटर बनाएं

अब जब हमने एक परिदृश्य देखा है, तो चलिए उस परिदृश्य के अनुसार कोड लिखते हैं। ऐसा करने के लिए, ये चरण अपनाएं:

1. मौजूदा फाइल को शुरुआती बिंदु के रूप में इस्तेमाल करें
1. एक `prompt` वेरिएबल बनाएं और नीचे दिए गए अनुसार सैंपल कोड में बदलाव करें:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

बहुत बढ़िया, अब देखते हैं कि हम चीजों को कैसे बेहतर बना सकते हैं। चीजों को बेहतर बनाने के लिए, हम चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या को आसानी से सुधारा और बदला जा सके।


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
