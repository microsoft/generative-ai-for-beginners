# टेक्स्ट जनरेशन ऐप्स बनाएं

आपने अब तक इस पाठ्यक्रम में देखा है कि प्रम्प्ट्स जैसे मूलभूत अवधारणाएँ हैं और एक पूरी विधा "प्रम्प्ट इंजीनियरिंग" भी है। कई उपकरण जिनके साथ आप बातचीत कर सकते हैं जैसे ChatGPT, Office 365, Microsoft Power Platform और अधिक, आपको कुछ पूरा करने के लिए प्रम्प्ट्स का उपयोग करने का समर्थन करते हैं।

यदि आप ऐसी अनुभव एक ऐप में जोड़ना चाहते हैं, तो आपको प्रम्प्ट्स, पूर्णताएं जैसे अवधारणाओं को समझना होगा और काम करने के लिए एक पुस्तकालय चुननी होगी। यही आपको इस अध्याय में सीखने को मिलेगा।

## परिचय

इस अध्याय में, आप:

- openai लाइब्रेरी और इसकी मूल अवधारणाओं के बारे में जानेंगे।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएंगे।
- टेक्स्ट जनरेशन ऐप बनाने के लिए प्रम्प्ट, तापमान, और टोकन जैसी अवधारणाओं का उपयोग कैसे करें, समझेंगे।

## सीखने के उद्देश्य

इस पाठ के अंत में, आप सक्षम होंगे:

- बताएं कि टेक्स्ट जनरेशन ऐप क्या है।
- openai का उपयोग करके एक टेक्स्ट जनरेशन ऐप बनाएं।
- अपने ऐप को अधिक या कम टोकन के लिए कॉन्फ़िगर करें और आउटपुट विविधता के लिए तापमान भी बदलें।

## टेक्स्ट जनरेशन ऐप क्या है?

सामान्य तौर पर जब आप ऐप बनाते हैं तो उसमें कुछ प्रकार का इंटरफ़ेस होता है जैसे नीचे दिया गया:

- कमांड-आधारित। कंसोल ऐप्स वे ऐप्स होते हैं जहां आप कमांड टाइप करते हैं और यह कार्य करता है। उदाहरण के लिए, `git` एक कमांड-आधारित ऐप है।
- उपयोगकर्ता इंटरफ़ेस (UI)। कुछ ऐप्स में ग्राफिकल यूजर इंटरफेस (GUIs) होते हैं जिनमें आप बटन क्लिक करते हैं, टेक्स्ट इनपुट करते हैं, विकल्प चुनते हैं, आदि।

### कंसोल और UI ऐप सीमित होते हैं

इसे एक कमांड-आधारित ऐप से तुलना करें जहां आप कमांड टाइप करते हैं:

- **यह सीमित है**। आप कोई भी कमांड नहीं टाइप कर सकते, केवल वे जो ऐप समर्थन करता है।
- **भाषा-विशिष्ट**। कुछ ऐप कई भाषाओं का समर्थन करते हैं, लेकिन डिफ़ॉल्ट रूप से ऐप एक विशिष्ट भाषा के लिए बनाया गया होता है, भले ही आप अधिक भाषा समर्थन जोड़ सकें।

### टेक्स्ट जनरेशन ऐप के लाभ

तो टेक्स्ट जनरेशन ऐप कैसे अलग है?

एक टेक्स्ट जनरेशन ऐप में आपको अधिक लचीलापन मिलता है, आप कमांड सेट या विशेष इनपुट भाषा तक सीमित नहीं होते। इसके बजाय, आप प्राकृतिक भाषा का उपयोग करके ऐप से बातचीत कर सकते हैं। एक और लाभ यह है कि आप पहले से ही एक ऐसे डेटा स्रोत के साथ बातचीत कर रहे हैं जिसे विशाल सूचना संग्रह पर प्रशिक्षित किया गया है, जबकि पारंपरिक ऐप एक डेटाबेस में सीमित हो सकता है।

### टेक्स्ट जनरेशन ऐप से मैं क्या बना सकता हूँ?

आप कई चीजें बना सकते हैं। उदाहरण के लिए:

- **एक चैटबॉट**। एक चैटबॉट जो आपके कंपनी और उसके उत्पादों से संबंधित प्रश्नों का उत्तर देता है, अच्छा मेल हो सकता है।
- **सहायक**। LLM टेक्स्ट सारांशित करने, टेक्स्ट से अंतर्दृष्टि प्राप्त करने, रिज़्यूमे जैसे टेक्स्ट उत्पन्न करने में बहुत अच्छे हैं।
- **कोड सहायक**। आपकी उपयोग की जाने वाली भाषा मॉडल पर निर्भर करता है, आप कोड लिखने में मदद करने वाला कोड सहायक बना सकते हैं। उदाहरण के लिए, आप GitHub Copilot और ChatGPT जैसे उत्पादों का उपयोग कोड लिखने में सहायता के लिए कर सकते हैं।

## मैं कैसे शुरू कर सकता हूँ?

ठीक है, आपको LLM के साथ एकीकृत करने का तरीका खोजना होगा जो आमतौर पर निम्न दो तरीकों से होता है:

- API का उपयोग करें। यहां आप अपने प्रम्प्ट के साथ वेब अनुरोध बनाते हैं और उत्पन्न टेक्स्ट प्राप्त करते हैं।
- पुस्तकालय का उपयोग करें। पुस्तकालय API कॉल्स को संक्षिप्त करते हैं और उनका उपयोग आसान बनाते हैं।

## पुस्तकालय/SDKs

LLM के साथ काम करने के लिए कुछ प्रसिद्ध पुस्तकालय हैं जैसे:

- **openai**, यह लाइब्रेरी आपके मॉडल से कनेक्ट करना और प्रम्प्ट भेजना आसान बनाती है।

फिर कुछ उच्च स्तर के पुस्तकालय हैं जैसे:

- **Langchain**। Langchain प्रसिद्ध है और Python का समर्थन करता है।
- **Semantic Kernel**। Semantic Kernel Microsoft द्वारा बनाई गई एक लाइब्रेरी है जो C#, Python, और Java भाषाओं का समर्थन करती है।

## ChatCompletions के साथ टेक्स्ट जेनरेट करें

टेक्स्ट जेनरेट करने का तरीका `ChatCompletionsClient` क्लास का उपयोग करना है।
`samples/python/azure_ai_inference/basic.py` में, प्रतिक्रिया कोड अनुभाग में, user role के content पैरामीटर को नीचे दिए अनुसार अपडेट करें:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ऐच्छिक पैरामीटर
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट की गई फ़ाइल चलाएँ और आउटपुट देखें


## अलग-अलग चीज़ों के लिए अलग-अलग प्रकार के प्रॉम्प्ट

अब आपने देखा है कि प्रॉम्प्ट का उपयोग करके टेक्स्ट कैसे जनरेट किया जाता है। आपके पास एक प्रोग्राम भी चल रहा है जिसे आप अलग-अलग प्रकार के टेक्स्ट जनरेट करने के लिए संशोधित और बदल सकते हैं।  

प्रॉम्प्ट का उपयोग सभी तरह के कार्यों के लिए किया जा सकता है। उदाहरण के लिए:

- **टेक्स्ट का एक प्रकार जनरेट करें**। उदाहरण के लिए, आप एक कविता, क्विज़ के लिए प्रश्न आदि जनरेट कर सकते हैं।
- **जानकारी खोजें**। आप प्रॉम्प्ट का उपयोग जानकारी खोजने के लिए कर सकते हैं, जैसे कि निम्न उदाहरण: 'वेब डेवलपमेंट में CORS का क्या मतलब है?'।
- **कोड जनरेट करें**। आप प्रॉम्प्ट का उपयोग कोड जनरेट करने के लिए कर सकते हैं, उदाहरण के लिए ईमेल वैधता के लिए उपयुक्त रेगुलर एक्सप्रेशन विकसित करना या पूरा प्रोग्राम जनरेट करना, जैसे कि एक वेब ऐप?

## अभ्यास - एक रेसिपी जनरेटर बनाएं

अब जब हम एक परिदृश्य का अभ्यास कर चुके हैं, तो चलिए उस प्रदर्शित परिदृश्य से मेल खाने वाला कोड लिखते हैं। ऐसा करने के लिए, निम्न चरणों का पालन करें:

1. मौजूदा फ़ाइल को एक प्रारंभिक बिंदु के रूप में उपयोग करें
1. एक `prompt` वेरिएबल बनाएं और नमूना कोड को नीचे दिए अनुसार बदलें:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

बढ़िया, आइए देखें कि हम चीज़ों को कैसे बेहतर बना सकते हैं। चीज़ों को बेहतर बनाने के लिए, हम यह सुनिश्चित करना चाहते हैं कि कोड लचीला हो, ताकि सामग्री और रेसिपी की संख्या को सुधारा और बदला जा सके।


1. आइए कोड को निम्नलिखित तरीके से बदलें:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **तापमान के साथ प्रयोग**। तापमान ऐसी चीज है जिसके बारे में हमने अब तक उल्लेख नहीं किया है लेकिन यह हमारे कार्यक्रम के प्रदर्शन के लिए महत्वपूर्ण संदर्भ है। तापमान मान जितना अधिक होगा, आउटपुट उतना ही यादृच्छिक होगा। इसके विपरीत, तापमान मान जितना कम होगा, आउटपुट उतना ही पूर्वानुमानित होगा। विचार करें कि क्या आप अपने आउटपुट में विविधता चाहते हैं या नहीं।

   तापमान बदलने के लिए, आप `temperature` पैरामीटर का उपयोग कर सकते हैं। उदाहरण के लिए, यदि आप 0.5 का तापमान उपयोग करना चाहते हैं, तो आप ऐसा करेंगे:

```python
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
            },
            {
                "role": "user",
                "content": new_prompt,
            },
        ],
        model=model_name,
        # वैकल्पिक पैरामीटर
        temperature=0.5,
        max_tokens=1200,
        top_p=1.    
    )
```

   > ध्यान दें, 1.0 के करीब होने पर आउटपुट अधिक विविध होगा।


## असाइनमेंट

इस असाइनमेंट के लिए, आप चुन सकते हैं कि क्या बनाना है।

यहां कुछ सुझाव हैं:

- रेसिपी जनरेटर ऐप को और सुधारने के लिए ट्वीक करें। तापमान मानों और प्रॉम्प्ट के साथ खेलें और देखें कि आप क्या बना सकते हैं।
- एक "स्टडी बड्डी" बनाएं। यह ऐप किसी विषय जैसे पाइथन के बारे में प्रश्नों का उत्तर देने में सक्षम होना चाहिए, आपके पास ऐसे प्रॉम्प्ट हो सकते हैं जैसे "पाइथन में किसी निश्चित विषय का क्या मतलब है?", या आप ऐसा प्रॉम्प्ट दे सकते हैं जो कहता है, मुझे किसी निश्चित विषय के लिए कोड दिखाओ आदि।
- इतिहास बॉट, इतिहास को जीवंत करें, बॉट को एक विशिष्ट ऐतिहासिक चरित्र का अभिनय करने का निर्देश दें और उसके जीवन और समय के बारे में प्रश्न पूछें।
