## O que é uma aplicação de geração de texto?

Normalmente, quando constróis uma aplicação, ela tem algum tipo de interface como a seguinte:

- Baseada em comandos. Aplicações de consola são aplicações típicas onde escreves um comando e ela executa uma tarefa. Por exemplo, `git` é uma aplicação baseada em comandos.
- Interface de utilizador (UI). Algumas aplicações têm interfaces gráficas (GUIs) onde clicas em botões, escreves texto, selecionas opções e mais.

### Aplicações de consola e UI são limitadas

Compara com uma aplicação baseada em comandos onde escreves um comando: 

- **É limitado**. Não podes simplesmente escrever qualquer comando, só os que a aplicação suporta.
- **Específico para linguagem**. Algumas aplicações suportam muitas linguagens, mas por defeito a aplicação é construída para uma linguagem específica, mesmo que possas adicionar suporte para mais linguagens. 

### Benefícios das aplicações de geração de texto

Então, como é que uma aplicação de geração de texto é diferente?

Numa aplicação de geração de texto, tens mais flexibilidade, não estás limitado a um conjunto de comandos ou a uma linguagem de input específica. Em vez disso, podes usar linguagem natural para interagir com a aplicação. Outro benefício é que, porque já estás a interagir com uma fonte de dados que foi treinada num vasto corpus de informação, ao contrário de uma aplicação tradicional que pode estar limitada ao que há numa base de dados. 

### O que posso construir com uma aplicação de geração de texto?

Há muitas coisas que podes construir. Por exemplo:

- **Um chatbot**. Um chatbot que responde a perguntas sobre temas, como a tua empresa e os seus produtos, pode ser uma boa opção.
- **Auxiliar**. LLMs são ótimos em coisas como resumir texto, obter insights de texto, produzir texto como currículos e muito mais.
- **Assistente de código**. Dependendo do modelo de linguagem que usas, podes construir um assistente de código que te ajuda a escrever código. Por exemplo, podes usar um produto como GitHub Copilot assim como o ChatGPT para ajudar a escrever código.

## Gerar texto

A forma de gerar texto é usar o método `responses.create`. Aqui está um exemplo:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

No código acima, criamos um objeto de resposta e passamos o modelo que queremos usar e o prompt. Depois mostramos o texto gerado.

### Respostas em chat

Até agora, viste como podemos gerar texto a partir de um único prompt. A API Responses também é adequada para chatbots - forneces uma lista de mensagens como `input` para construir uma conversa. Aqui está um exemplo do seu uso:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## Exercício - criar um gerador de receitas

Agora que representámos um cenário, vamos escrever código para corresponder ao cenário demonstrado. Para isso, siga os seguintes passos:

1. Use o ficheiro *app.py* existente como ponto de partida
1. Localize a variável `prompt` e altere o seu código para o seguinte:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. Vamos alterar o código da seguinte forma:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Produzir uma lista de compras**. Queremos produzir uma lista de compras, considerando o que já temos em casa.

    Para esta funcionalidade, poderíamos tentar resolver tudo num único prompt ou dividir em dois prompts. Vamos tentar a última abordagem. Aqui sugerimos adicionar um prompt adicional, mas para isso funcionar, precisamos adicionar o resultado do primeiro prompt como contexto ao segundo prompt.

    Localize a parte do código que imprime o resultado do primeiro prompt e adicione o seguinte código abaixo:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # imprimir resposta
    print("Shopping list:")
    print(response.output_text)
    ```

    Note o seguinte:

    - Estamos a construir um novo prompt adicionando o resultado do primeiro prompt ao novo prompt:
    
        ```python

        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Fazemos um novo pedido, mas considerando também o número de tokens pedidos no primeiro prompt, por isso desta vez dizemos que `max_output_tokens` é 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```

- **Uma palavra sobre o comprimento dos tokens**. Devemos considerar quantos tokens precisamos para gerar o texto que queremos. Os tokens custam dinheiro, por isso, onde possível, devemos ser económicos com o número de tokens usados. Por exemplo, podemos formular o prompt de forma a usar menos tokens?

   Para alterar os tokens usados, pode-se usar o parâmetro `max_output_tokens`. Por exemplo, se quiser usar 100 tokens, faria:


- **Experimentar com a temperatura**. A temperatura é algo que ainda não mencionámos, mas é um contexto importante para o desempenho do nosso programa. Quanto maior o valor da temperatura, mais aleatória será a saída. Pelo contrário, quanto menor a temperatura, mais previsível será a saída. Considere se quer variação na sua saída ou não.

   Para alterar a temperatura, pode usar o parâmetro `temperature`. Por exemplo, se quiser usar uma temperatura de 0.5, faria:


   > Nota, quanto mais próximo de 1.0, mais variada será a saída.
