## What is a text generation app?

Usually, when you build an app, it has some kind of interface like:

- Command-based. Console apps are classic examples where you type a command and it does something. For instance, `git` is a command-based app.
- User interface (UI). Some apps have graphical user interfaces (GUIs) where you click buttons, enter text, select options, and more.

### Console and UI apps are limited

Compare this to a command-based app where you type a command:

- **It's limited**. You can't just type anything; only the commands the app supports will work.
- **Language specific**. Some apps support multiple languages, but usually they're built for one language, even if you can add more later.

### Benefits of text generation apps

So, what makes a text generation app different?

With a text generation app, you have more freedom—you're not stuck with a fixed set of commands or a specific input language. You can use natural language to interact with the app. Another advantage is that you're working with a data source trained on a huge amount of information, while a traditional app might only have what's in its database.

### What can I build with a text generation app?

There are lots of possibilities. For example:

- **A chatbot**. You could build a chatbot that answers questions about your company and its products.
- **Helper**. LLMs are great for things like summarizing text, finding insights, generating resumes, and more.
- **Code assistant**. Depending on the language model, you can create a code assistant to help you write code. Tools like GitHub Copilot and ChatGPT can help with this.

## How can I get started?

To get started, you need a way to connect to an LLM, which usually means one of two approaches:

- Use an API. You send web requests with your prompt and get generated text back.
- Use a library. Libraries wrap the API calls and make them easier to use.

## Libraries/SDKs

There are some well-known libraries for working with LLMs, such as:

- **openai**, which makes it simple to connect to your model and send prompts.

There are also higher-level libraries like:

- **Langchain**. Langchain is popular and works with Python.
- **Semantic Kernel**. Semantic Kernel is a Microsoft library that supports C#, Python, and Java.

## Exercise - build a recipe generator

Now that we've gone through a scenario, let's write code that matches what we just saw. To do this, follow these steps:

1. Use the current file as your starting point
1. Create a `prompt` variable and update the sample code as shown below:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Great, let's see how we can improve things. To improve things, we want to make sure the code is flexible, so ingredients and number of recipes can be improved and changed.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

```python
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
```
