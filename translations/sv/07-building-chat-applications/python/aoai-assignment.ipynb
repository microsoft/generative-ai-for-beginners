### 2. Importera hjälpbibliotek och skapa instanser av autentiseringsuppgifter


```python
import os
from openai import OpenAI
import numpy as np
from dotenv import load_dotenv
load_dotenv()

#validate data inside .env file

# The Responses API is served from the Azure OpenAI (Microsoft Foundry) v1 endpoint,
# so we point the OpenAI client at <your-endpoint>/openai/v1/ (no api_version needed).
endpoint = os.environ['AZURE_OPENAI_ENDPOINT']
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{endpoint.rstrip('/')}/openai/v1/",
  )
```

### 3. Hitta rätt modell  
Modeller som GPT-4o och GPT-4o mini kan förstå och generera naturligt språk. Azure OpenAI i Microsoft Foundry erbjuder en rad modellkapaciteter, var och en med olika nivåer av kraft och hastighet som passar för olika uppgifter. 

[Azure OpenAI in Microsoft Foundry models](https://learn.microsoft.com/azure/ai-foundry/foundry-models/concepts/models-sold-directly-by-azure?WT.mc_id=academic-105485-koreyst)



```python
# Select the deployment name configured in your .env file
model = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

#### Hur man kontrollerar en stor språkmodell  
"Av alla insignal till en stor språkmodell är textprompten(1) klart den mest inflytelserika.

Stora språkmodeller kan prompts för att producera output på några sätt:

- Instruktion: Berätta för modellen vad du vill ha
- Avslut: Få modellen att färdigställa början på det du vill ha
- Demonstration: Visa modellen vad du vill ha, med antingen:
- Några få exempel i prompten
- Många hundratals eller tusentals exempel i en finjusterings-träningsdataset



#### Det finns tre grundläggande riktlinjer för att skapa prompts:

**Visa och berätta**. Gör det tydligt vad du vill ha, antingen genom instruktioner, exempel eller en kombination av båda. Om du vill att modellen ska rangordna en lista av objekt i alfabetisk ordning eller klassificera ett stycke efter sentiment, visa att det är vad du vill ha.

**Ge kvalitetsdata**. Om du försöker bygga en klassificerare eller få modellen att följa ett mönster, se till att det finns tillräckligt med exempel. Se till att korrekturläsa dina exempel — modellen är vanligtvis smart nog att genomskåda grundläggande stavfel och ge dig ett svar, men den kan också anta att detta är avsiktligt och det kan påverka svaret.

**Kontrollera dina inställningar.** Inställningarna temperature och top_p styr hur deterministisk modellen är i att generera ett svar. Om du ber om ett svar där det bara finns ett rätt svar vill du sätta dessa lägre. Om du vill ha mer varierande svar kan du vilja ställa in dem högre. Det vanligaste misstaget folk gör med dessa inställningar är att anta att de är regler för "klokhet" eller "kreativitet".

### 5. Skicka in!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Inbäddningar
Den här sektionen visar hur man hämtar inbäddningar och hittar likheter mellan ord, meningar och dokument. För att köra följande anteckningsböcker behöver du distribuera en modell som använder `text-embedding-ada-002` som basmodell och ange dess distributionsnamn i .env-filen med variabeln `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT`.


### Modelltaxonomi – Välja en inbäddningsmodell

**Modelltaxonomi**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (inbäddningsfamilj)  
{capability} --> ada             (alla andra inbäddningsmodeller kommer att fasas ut 2024)  
{input-type} --> n/a             (anges endast för sökmodeller)  
{identifier} --> 002             (version 002)  

model = 'text-embedding-ada-002'


  > [!NOTE] Följande steg är inte nödvändigt om du kör den här anteckningsboken på Codespaces eller inom en Devcontainer


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
