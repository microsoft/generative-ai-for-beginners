# पाठ उत्पादन एपहरू बनाउनुहोस्

यस पाठ्यक्रममा अहिलेसम्म तपाईंले प्रॉम्प्ट जस्ता मुख्य अवधारणाहरू देख्नुभएको छ, र "प्रॉम्प्ट इन्जिनियरिङ" नामको छुट्टै विधा पनि छ। तपाईंले प्रयोग गर्न सक्ने धेरै उपकरणहरू जस्तै ChatGPT, Office 365, Microsoft Power Platform आदिले तपाईंलाई प्रॉम्प्ट प्रयोग गरेर केही गर्न सहयोग गर्छन्।

यस्तो अनुभव तपाईंको एपमा थप्नका लागि, तपाईंले प्रॉम्प्ट, कम्प्लिशन जस्ता अवधारणाहरू बुझ्नुपर्छ र कुन लाइब्रेरी प्रयोग गर्ने भन्ने निर्णय गर्नुपर्छ। यही कुरा तपाईंले यस अध्यायमा सिक्नुहुनेछ।

## परिचय

यस अध्यायमा, तपाईंले:

- openai लाइब्रेरी र यसको मुख्य अवधारणाहरूको बारेमा जान्नुहुनेछ।
- openai प्रयोग गरेर पाठ उत्पादन एप बनाउनुहुनेछ।
- प्रॉम्प्ट, टेम्परेचर, र टोकन जस्ता अवधारणाहरू प्रयोग गरेर पाठ उत्पादन एप कसरी बनाउने भन्ने कुरा बुझ्नुहुनेछ।

## सिक्ने लक्ष्यहरू

यस पाठको अन्त्यमा, तपाईंले:

- पाठ उत्पादन एप के हो भन्ने व्याख्या गर्न सक्नुहुनेछ।
- openai प्रयोग गरेर पाठ उत्पादन एप बनाउनुहुनेछ।
- तपाईंको एपमा बढी वा कम टोकन प्रयोग गर्ने र टेम्परेचर परिवर्तन गर्ने जस्ता सेटिङहरू गर्न सक्नुहुनेछ, ताकि नतिजा फरक-फरक आउन सकोस्।

## पाठ उत्पादन एप के हो?

सामान्यतया जब तपाईं एप बनाउनुहुन्छ, त्यसमा तलका जस्ता इन्टरफेस हुन्छन्:

- कमाण्ड-आधारित। कन्सोल एपहरूमा तपाईंले कमाण्ड टाइप गर्नुहुन्छ र त्यो कार्य गर्छ। उदाहरणका लागि, `git` एउटा कमाण्ड-आधारित एप हो।
- युजर इन्टरफेस (UI)। केही एपहरूमा ग्राफिकल युजर इन्टरफेस (GUI) हुन्छ, जहाँ तपाईंले बटन क्लिक गर्नुहुन्छ, टेक्स्ट इनपुट गर्नुहुन्छ, विकल्पहरू छान्नुहुन्छ आदि।

### कन्सोल र UI एपहरू सीमित छन्

कमाण्ड-आधारित एपसँग तुलना गर्दा:

- **यो सीमित छ।** तपाईंले जुनसुकै कमाण्ड टाइप गर्न सक्नुहुन्न, एपले समर्थन गरेका कमाण्डहरू मात्र प्रयोग गर्न सक्नुहुन्छ।
- **भाषा-विशेष।** केही एपहरूले धेरै भाषा समर्थन गर्छन्, तर डिफल्टमा एप कुनै एक भाषाका लागि बनाइएको हुन्छ, यद्यपि थप भाषा समर्थन गर्न सकिन्छ।

### पाठ उत्पादन एपका फाइदाहरू

त्यसो भए पाठ उत्पादन एप कसरी फरक छ?

पाठ उत्पादन एपमा तपाईंलाई बढी स्वतन्त्रता हुन्छ, तपाईं कमाण्डहरूको सेट वा कुनै विशेष इनपुट भाषामा सीमित हुनुहुन्न। बरु, तपाईंले प्राकृतिक भाषा प्रयोग गरेर एपसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा के छ भने, तपाईंले पहिले नै ठूलो डाटासेटमा तालिम लिएको स्रोतसँग अन्तरक्रिया गर्दै हुनुहुन्छ, जबकि परम्परागत एपमा डाटाबेसमा भएको जानकारीमा मात्र निर्भर हुन्छ।

### पाठ उत्पादन एपबाट के बनाउन सकिन्छ?

धेरै कुरा बनाउन सकिन्छ। उदाहरणका लागि:

- **च्याटबोट।** कम्पनी र यसको उत्पादनबारे प्रश्नहरूको उत्तर दिने च्याटबोट राम्रो विकल्प हुन सक्छ।
- **सहायक।** LLM हरू टेक्स्ट समरी गर्न, टेक्स्टबाट जानकारी निकाल्न, रिजुमे जस्ता टेक्स्ट उत्पादन गर्न आदि काममा उत्कृष्ट छन्।
- **कोड असिस्टेन्ट।** तपाईंले जुन भाषा मोडेल प्रयोग गर्नुहुन्छ, त्यसअनुसार कोड लेख्न सहयोग गर्ने असिस्टेन्ट बनाउन सक्नुहुन्छ। उदाहरणका लागि, GitHub Copilot वा ChatGPT प्रयोग गरेर कोड लेख्न सहयोग लिन सकिन्छ।

## कसरी सुरु गर्ने?

तपाईंले LLM सँग कसरी जोड्ने भन्ने उपाय खोज्नुपर्छ, सामान्यतया दुई तरिका छन्:

- API प्रयोग गर्नुहोस्। यहाँ तपाईंले वेब अनुरोध बनाउनुहुन्छ, प्रॉम्प्ट पठाउनुहुन्छ र उत्पन्न भएको टेक्स्ट फर्किन्छ।
- लाइब्रेरी प्रयोग गर्नुहोस्। लाइब्रेरीहरूले API कलहरूलाई सजिलो बनाउँछन्।

## लाइब्रेरीहरू/SDK हरू

LLM सँग काम गर्नका लागि केही चर्चित लाइब्रेरीहरू छन्:

- **openai**, यो लाइब्रेरीले तपाईंको मोडेलसँग सजिलै जडान गर्न र प्रॉम्प्ट पठाउन सजिलो बनाउँछ।

त्यसपछि केही उच्च स्तरका लाइब्रेरीहरू छन्:

- **Langchain**। Langchain प्रख्यात छ र Python समर्थन गर्छ।
- **Semantic Kernel**। Semantic Kernel Microsoft को लाइब्रेरी हो, जसले C#, Python, र Java समर्थन गर्छ।

अब तपाईंले `openai` लाइब्रेरीलाई कुन API key प्रयोग गर्ने भन्ने बताउनुपर्छ। तपाईंको API key पत्ता लगाउन, Azure OpenAI स्रोतको "Keys and Endpoint" सेक्सनमा जानुहोस् र "Key 1" को मान कपी गर्नुहोस्।

अब तपाईंले यो जानकारी कपी गर्नुभयो भने, लाइब्रेरीहरूलाई प्रयोग गर्न निर्देशन दिनुहोस्।

> [!NOTE]
> तपाईंको API key लाई कोडबाट छुट्टै राख्नु राम्रो हुन्छ। तपाईंले environment variable प्रयोग गरेर गर्न सक्नुहुन्छ।
> - .env फाइलमा `AZURE_OPENAI_API_KEY` environment variable सेट गर्नुहोस्। यदि तपाईंले यस कोर्सका अघिल्ला अभ्यासहरू पूरा गर्नुभएको छ भने, तपाईं तयार हुनुहुन्छ।

### Azure को कन्फिगरेसन सेटअप गर्नुहोस्

यदि तपाईं Azure OpenAI प्रयोग गर्दै हुनुहुन्छ भने, कन्फिगरेसन यसरी सेटअप गर्नुहोस्:

```python
client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT')
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

माथि हामीले तलका कुरा सेट गरेका छौं:

- `api_key`, यो तपाईंको Azure Portal मा भेटिने API key हो।
- `api_version`, तपाईंले प्रयोग गर्न चाहेको API को संस्करण हो। यो लेख्दा, नयाँ संस्करण `2023-10-01-preview` हो।
- `azure_endpoint`, API को endpoint हो। Azure Portal मा तपाईंको API key को छेउमा पाउन सकिन्छ।

> [!NOTE]
> `os.environ` environment variable पढ्ने function हो। तपाईंले environment variable जस्तै `AZURE_OPENAI_API_KEY` र `AZURE_OPENAI_ENDPOINT` पढ्न प्रयोग गर्न सक्नुहुन्छ।

## पाठ उत्पादन गर्नुहोस्

पाठ उत्पादन गर्ने तरिका `chat.completion` class प्रयोग गर्नु हो। तल उदाहरण छ:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

माथिको कोडमा, हामीले completion object बनायौं र प्रयोग गर्न चाहेको मोडेल र प्रॉम्प्ट पास गर्यौं। त्यसपछि उत्पन्न भएको टेक्स्ट प्रिन्ट गर्यौं।

### च्याट कम्प्लिशनहरू

अहिलेसम्म तपाईंले `Completion` प्रयोग गरेर पाठ उत्पादन गर्ने देख्नुभयो। तर अर्को class छ, `ChatCompletion`, जुन च्याटबोटका लागि बढी उपयुक्त छ। यसलाई प्रयोग गर्ने उदाहरण:

```python
client = AzureOpenAI(
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT'), 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-05-15"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

यसको बारेमा थप जानकारी आउने अध्यायमा हुनेछ।

## अभ्यास - तपाईंको पहिलो पाठ उत्पादन एप

अब हामीले Azure OpenAI सेवा सेटअप र कन्फिगर गर्ने सिक्यौं, अब तपाईंको पहिलो पाठ उत्पादन एप बनाउने समय आएको छ। एप बनाउनका लागि तलका चरणहरू पालना गर्नुहोस्:


1. भर्चुअल वातावरण बनाउनुहोस् र openai इन्स्टल गर्नुहोस्:

  > [!NOTE] यदि तपाईंले यो नोटबुक Codespaces वा Devcontainer भित्र चलाउनुहुन्छ भने यो चरण आवश्यक छैन

> [!NOTE]

> [!NOTE]
> आफ्नो Azure OpenAI कुञ्जी पत्ता लगाउन https://portal.azure.com/ मा जानुहोस् र `Open AI` खोज्नुहोस्, त्यसपछि `Open AI resource` चयन गर्नुहोस्, अनि `Keys and Endpoint` मा जानुहोस् र `Key 1` को मान प्रतिलिपि गर्नुहोस्।


१. *app.py* फाइल बनाउनुहोस् र यसमा तलको कोड राख्नुहोस्:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

तपाईंले तलको जस्तो नतिजा देख्नु पर्नेछ:

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## फरक-फरक कुराका लागि फरक-फरक प्रकारका प्रम्प्टहरू

अब तपाईंले प्रम्प्ट प्रयोग गरेर कसरी पाठ उत्पन्न गर्ने देख्नुभयो। तपाईंले चलिरहेको एउटा प्रोग्राम पनि पाउनुभएको छ जसलाई तपाईंले परिवर्तन गरेर विभिन्न प्रकारका पाठहरू बनाउन सक्नुहुन्छ।

प्रम्प्टहरू धेरै किसिमका कामका लागि प्रयोग गर्न सकिन्छ। उदाहरणका लागि:

- **पाठको कुनै प्रकार उत्पन्न गर्नुहोस्**। जस्तै, तपाईंले कविता, क्विजका लागि प्रश्नहरू आदि बनाउन सक्नुहुन्छ।
- **जानकारी खोज्नुहोस्**। तपाईंले प्रम्प्ट प्रयोग गरेर जानकारी खोज्न सक्नुहुन्छ, जस्तै 'वेब विकासमा CORS भन्नाले के जनाउँछ?'।
- **कोड उत्पन्न गर्नुहोस्**। तपाईंले प्रम्प्ट प्रयोग गरेर कोड बनाउन सक्नुहुन्छ, जस्तै इमेल जाँच गर्न प्रयोग हुने रेगुलर एक्सप्रेसन बनाउने, वा किन न पूरै प्रोग्राम नै बनाउने, जस्तै वेब एप?

1. पहिलेबाट रहेको *app.py* फाइललाई सुरुवातको आधारको रूपमा प्रयोग गर्नुहोस्
1. `prompt` भेरिएबल खोज्नुहोस् र यसको कोडलाई तलको अनुसार परिवर्तन गर्नुहोस्:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

राम्रो, अब हेरौं कसरी हामीले यसलाई अझ राम्रो बनाउन सक्छौं। सुधार गर्नका लागि, हामी चाहन्छौं कि कोड लचिलो होस्, ताकि सामग्रीहरू र रेसिपीहरूको संख्या सजिलै परिवर्तन र सुधार गर्न सकियोस्।


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
