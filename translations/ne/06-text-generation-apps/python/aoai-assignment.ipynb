# टेक्स्ट जेनरेशन एपहरू बनाउनुहोस्

तपाईले अहिलेसम्म यो पाठ्यक्रम मार्फत देख्नुभएको छ कि प्रॉम्प्टहरू जस्ता मूल अवधारणाहरू छन् र यहाँ सम्म कि पूर्ण रूपमा "प्रॉम्प्ट इन्जिनियरिङ" नामक एक विषय पनि छ। धेरै उपकरणहरू जुन तपाईं अन्तरक्रिया गर्न सक्नुहुन्छ जस्तै ChatGPT, Office 365, Microsoft Power Platform र थप, तपाईलाई प्रॉम्प्टहरू प्रयोग गरेर केहि गर्न सहयोग गर्छन्।

तपाईले यस्ता अनुभवलाई कुनै एपमा थप्न चाहनु हुन्छ भने, तपाईलाई प्रॉम्प्ट, कम्प्लीसन जस्ता अवधारणाहरू बुझ्नुपर्छ र काम गर्नको लागि एउटा लाइब्रेरी छनौट गर्नुपर्छ। यही कुरा तपाईं यस अध्यायमा सिक्नु हुनेछ।

## परिचय

यस अध्यायमा, तपाईं:

- openai लाइब्रेरी र यसको मूल अवधारणाहरूको बारेमा जान्नुहोस्।
- openai प्रयोग गरेर एउटा टेक्स्ट जेनरेशन एप बनाउनुहोस्।
- प्रॉम्प्ट, तापक्रम, र टोकन्स जस्ता अवधारणाहरू कसरी प्रयोग गर्ने जान्नुहोस् ताकि टेक्स्ट जेनरेशन एप बनाउनु सकियोस्।

## सिकाइ लक्ष्यहरू

यस पाठको अन्त्यमा, तपाईं सक्षम हुनुहुनेछ:

- टेक्स्ट जेनरेशन एप के हो भन्ने व्याख्या गर्नुहोस्।
- openai प्रयोग गरेर टेक्स्ट जेनरेशन एप बनाउनुहोस्।
- आफ्नो एपलाई बढी वा कम टोकन्स प्रयोग गर्न र तापक्रम परिवर्तन गर्न कन्फिगर गर्नुहोस्, फरक फरक परिणामको लागि।

## टेक्स्ट जेनरेशन एप के हो?

सामान्यतया एप बनाउँदा यसले केही प्रकारको इन्टरफेस हुन्छ जस्तै तल दिइएको:

- कमाण्ड-आधारित। कन्सोल एपहरू सामान्य एपहरू हुन् जहाँ तपाईं कमाण्ड टाइप गर्नुहुन्छ र एपले कार्य सम्पन्न गर्छ। उदाहरणका लागि, `git` एक कमाण्ड-आधारित एप हो।
- प्रयोगकर्ता इन्टरफेस (UI)। कतिपय एपहरूमा ग्राफिकल युजर इन्टरफेस (GUI) हुन्छ जहाँ तपाईं बटन क्लिक गर्नुहुन्छ, टेक्स्ट इनपुट गर्नुहुन्छ, विकल्पहरू चयन गर्नुहुन्छ र थप।

### कन्सोल र UI एपहरू सीमित छन्

यसलाई कमाण्ड-आधारित एपसँग तुलना गर्नुहोस् जहाँ तपाईं कमाण्ड टाइप गर्नुहुन्छ: 

- **यो सीमित छ**। तपाईंले जुनसुकै कमाण्ड टाइप गर्न सक्नुहुँदैन, केवल ती जो एपले समर्थन गर्छ।
- **भाषा विशेष**। केही एपले धेरै भाषाहरू समर्थन गर्छन्, तर डिफल्टमा एप विशेष भाषा लागि बनाइएको हुन्छ, यद्यपि तपाईं थप भाषा समर्थन थप्न सक्नुहुन्छ।

### टेक्स्ट जेनरेशन एपहरूको फाइदा

टेक्स्ट जेनरेशन एप कसरी भिन्न छ?

टेक्स्ट जेनरेशन एपमा तपाईंलाई बढी लचीलोपना हुन्छ, तपाईं कमाण्डहरूको सेट वा विशिष्ट इनपुट भाषामा सीमित हुनुहुन्न। यसको सट्टा, तपाईं प्राकृतिक भाषाले एपसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा यो हो कि तपाईले यस्ता डाटास्रोतसँग पहिले नै अन्तरक्रिया गर्दै हुनुहुन्छ जुन विशाल जानकारीको कर्पसमा प्रशिक्षित छ, जबकि परम्परागत एप डेटाबेसमा सीमित हुन सक्छ।

### म टेक्स्ट जेनरेशन एपबाट के के बनाउन सक्छु?

धेरै कुरा बनाउन सकिन्छ। उदाहरणका लागि:

- **च्याटबोट**। कम्पनी र त्यसका उत्पादनहरूको बारेमा प्रश्नहरूको उत्तर दिने च्याटबोट राम्रो मेल खान सक्छ।
- **सहायक**। LLM हरू जस्ता कुराहरूमा उत्कृष्ट छन्, जस्तै टेक्स्ट सारांश, टेक्स्टबाट अन्वेषण प्राप्त गर्न, रेजुमे जस्ता टेक्स्ट उत्पादन गर्न।
- **कोड सहायक**। प्रयोग गरिने भाषा मोडेल अनुसार, तपाईं कोड लेख्न मद्दत गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, तपाईं GitHub Copilot जस्ता उत्पादन र ChatGPT दुवै प्रयोग गर्न सक्नुहुन्छ।

## म कसरी सुरु गर्न सक्छु?

ठीक छ, तपाईंले LLM सँग एकीकृत हुने तरिका खोज्नु पर्छ जुन सामान्यतया तलका दुई विधिहरू समेट्छ:

- API प्रयोग गर्नुहोस्। यहाँ तपाईं आफ्नो प्रॉम्प्टसहित वेब अनुरोध निर्माण गर्दै टेक्स्ट उत्पन्न प्राप्त गर्नुहुन्छ।
- लाइब्रेरी प्रयोग गर्नुहोस्। लाइब्रेरीहरूले API कलहरू समेट्छन् र प्रयोग गर्न सजिलो बनाउँछन्।

## लाइब्रेरीहरू/SDK हरू

LLM हरू सँग काम गर्ने केहि प्रख्यात लाइब्रेरीहरू छन् जस्तै:

- **openai**, यो लाइब्रेरीले तपाईंलाई मोडेलसँग जडान गर्न र प्रॉम्प्टहरू पठाउन सजिलो बनाउँछ।

त्यस पछि उच्च स्तरमा काम गर्ने लाइब्रेरीहरू पनि छन् जस्तै:

- **Langchain**। Langchain प्रख्यात छ र Python लाई समर्थन गर्छ।
- **Semantic Kernel**। Semantic Kernel माइक्रोसफ्टको लाइब्रेरी हो जुन C#, Python, र Java भाषाहरू समर्थन गर्छ।

### स्रोत सिर्जना गर्नुहोस्

यदि तपाईँले पहिले गरेन भने, तपाईंले तलका कदमहरू कार्यान्वयन गर्नुपर्छ:

- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) मा Azure OpenAI स्रोत सिर्जना गर्नुहोस्। कसरी [स्रोत सिर्जना गर्ने](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) जान्न यो मार्गनिर्देशन हेर्नुहोस्।


### API कुञ्जी र अन्त बिन्दु पत्ता लगाउनुहोस्

यस चरणमा, तपाईंले आफ्नो `openai` लाइब्रेरीलाई कुन API कुञ्जी प्रयोग गर्ने बताउनुपर्छ। API कुञ्जी पत्ता लगाउन, आफ्नो Azure OpenAI स्रोतको "Keys and Endpoint" सेक्सनमा जानुहोस् र "Key 1" मान प्रतिलिपि गर्नुहोस्।

  ![Azure Portal मा Keys and Endpoint स्रोत ब्लेड](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

अब तपाईंले यो जानकारी प्रतिलिपि गर्नुभएको छ, लाइब्रेरीहरूलाई यो प्रयोग गर्न निर्देशन दिऔं।

> [!NOTE]
> आफ्नो API कुञ्जीलाई तपाईंको कोडबाट अलग राख्नु राम्रो हुन्छ। यसका लागि वातावरण चरहरू प्रयोग गर्न सक्नुहुन्छ।
> - आफ्नो .env फाइलमा वातावरण चर `AZURE_OPENAI_API_KEY` मा आफ्नो API कुञ्जी सेट गर्नुहोस्। यदि तपाईंले यस कोर्सका अघिल्ला अभ्यासहरू पूरा गर्नु भएको छ भने, तपाईं सबै कुरा सेट गर्नुभएको छ।


### Azure संरचना सेटअप गर्नुहोस्

यदि तपाईं Azure OpenAI प्रयोग गर्दै हुनुहुन्छ भने, यहाँ कसरी संरचना सेटअप गर्ने हो। Responses API Azure OpenAI को **v1 endpoint** बाट सेवा दिइन्छ, त्यसैले हामीले `OpenAI` क्लाइन्टलाई `<your-endpoint>/openai/v1/` मा निर्देशित गर्छौं:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

माथि हामीले निम्न कुरा सेट गर्दैछौं:

- `api_key`, यो तपाईंको Azure Portal मा भेटिने API कुञ्जी हो।
- `base_url`, यो तपाईंको Azure OpenAI अन्त बिन्दु हो जुन `/openai/v1/` थपिएको छ। तपाईं अन्त बिन्दु Azure Portal मा आफ्नो API कुञ्जी नजिकै पाउन सक्नुहुन्छ। v1 endpoint प्रयोग गर्दा तपाईंले अब `api_version` पास गर्न आवश्यक छैन।
- `deployment`, यो Foundry पोर्टलमा बनाएको मोडेल डिप्लोयमेन्टको नाम हो।

> [!NOTE]
> `os.environ` वातावरण चरहरू पढ्ने म्यापिङ हो। यसले `AZURE_OPENAI_API_KEY` र `AZURE_OPENAI_ENDPOINT` जस्ता वातावरण चरहरू पढ्न मद्दत गर्छ।

## टेक्स्ट उत्पादन गर्नुहोस्

टेक्स्ट उत्पन्न गर्ने तरिका `responses.create` मेथड प्रयोग गर्नु हो। यहाँ एउटा उदाहरण छ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

माथिको कोडमा, हामीले एउटा प्रतिक्रिया वस्तु सिर्जना गर्छौं र प्रयोग गर्न चाहिने मोडेल र प्रॉम्प्ट पास गर्छौं। त्यसपछि हामी उत्पादन भएको टेक्स्ट प्रिन्ट गर्छौं।

### च्याट प्रतिक्रियाहरू

Responses API एकल चरण टेक्स्ट जेनरेशन र बहु चरण च्याटबोटहरू दुबैका लागि उपयुक्त छ - तपाईं केवल `input` सूचीमा थप संदेशहरू प्रदान गरेर कुराकानी निर्माण गर्नुहुन्छ:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

यस कार्यक्षमतामा आगामी अध्यायमा थप जानकारी दिइनेछ।

## अभ्यास - तपाईंको पहिलो टेक्स्ट जेनरेशन एप

अब हामीले Azure OpenAI सेवा सेट अप र कन्फिगर गर्ने तरिका सिकिसकेका छौं, तपाईंको पहिलो टेक्स्ट जेनरेशन एप बनाउने समय आयो। एप निर्माण गर्न यी कदमहरू पालना गर्नुहोस्:


1. एक भर्चुअल वातावरण सिर्जना गर्नुहोस् र openai स्थापना गर्नुहोस्:

  > [!NOTE] यदि तपाईंले यो नोटबुक Codespaces मा वा Devcontainer भित्र चलाउनु भयो भने यो चरण आवश्यक छैन


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> यदि तपाईं Windows प्रयोग गर्दै हुनुहुन्छ भने `source venv/bin/activate` को सट्टा `venv\Scripts\activate` टाइप गर्नुहोस्।

> [!NOTE]
> आफ्नो Azure OpenAI कुञ्जी पत्ता लगाउन https://portal.azure.com/ जानुहोस् र `Open AI` खोज्नुहोस् र `Open AI resource` चयन गर्नुहोस्, त्यसपछि `Keys and Endpoint` चयन गर्नुहोस् र `Key 1` को मान कपी गर्नुहोस्।


१. एउटा *app.py* फाइल बनाउनुहोस् र त्यसलाई निम्न कोड दिनुहोस्:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    तपाईंले निम्न जस्तै एक नतिजा देख्नु पर्छ:

    ```output
     अत्यन्तै असन्तुष्ट _____.

    एकपटकको कुरा हो, एक अत्यन्तै असन्तुष्ट जलपरी थिई।
    ```


## फरक कुराहरूका लागि फरक प्रकारका प्राम्प्टहरू

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
