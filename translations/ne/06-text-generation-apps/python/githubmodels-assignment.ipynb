# पाठ उत्पादन एपहरू बनाउनुहोस्

तपाईंले अहिलेसम्म यो पाठ्यक्रममा देख्नुभएको छ कि प्रॉम्प्ट जस्ता मुख्य अवधारणाहरू छन् र "प्रॉम्प्ट इन्जिनियरिङ" भन्ने पूरै एउटा विधा पनि छ। तपाईंले प्रयोग गर्न सक्ने धेरै उपकरणहरू जस्तै ChatGPT, Office 365, Microsoft Power Platform आदि, तपाईंलाई केही गर्न प्रॉम्प्ट प्रयोग गर्न समर्थन गर्छन्।

यस्तो अनुभव तपाईंको एपमा थप्नका लागि, तपाईंले प्रॉम्प्ट, कम्प्लिसन जस्ता अवधारणाहरू बुझ्नुपर्छ र कुन लाइब्रेरी प्रयोग गर्ने भन्ने निर्णय गर्नुपर्छ। यही कुरा तपाईंले यस अध्यायमा सिक्नुहुनेछ।

## परिचय

यस अध्यायमा, तपाईंले:

- openai लाइब्रेरी र यसको मुख्य अवधारणाहरूको बारेमा जान्नुहुनेछ।
- openai प्रयोग गरेर पाठ उत्पादन एप बनाउनुहुनेछ।
- प्रॉम्प्ट, टेम्परेचर, र टोकन जस्ता अवधारणाहरू प्रयोग गरेर कसरी पाठ उत्पादन एप बनाउने भन्ने बुझ्नुहुनेछ।

## सिकाइका लक्ष्यहरू

यस पाठको अन्त्यमा, तपाईंले:

- पाठ उत्पादन एप के हो भन्ने व्याख्या गर्न सक्नुहुनेछ।
- openai प्रयोग गरेर पाठ उत्पादन एप बनाउन सक्नुहुनेछ।
- तपाईंको एपमा बढी वा कम टोकन प्रयोग गर्न र टेम्परेचर परिवर्तन गर्न सक्नुहुनेछ, जसले गर्दा फरक-फरक नतिजा आउँछ।

## पाठ उत्पादन एप के हो?

सामान्यतया जब तपाईं एप बनाउनुहुन्छ, त्यसमा तलका जस्ता कुनै इन्टरफेस हुन्छ:

- कमाण्ड-आधारित। कन्सोल एपहरू त्यस्ता एपहरू हुन् जहाँ तपाईंले कमाण्ड टाइप गर्नुहुन्छ र त्यो कार्य गर्छ। उदाहरणका लागि, `git` एउटा कमाण्ड-आधारित एप हो।
- युजर इन्टरफेस (UI)। केही एपहरूमा ग्राफिकल युजर इन्टरफेस (GUI) हुन्छ जहाँ तपाईंले बटन क्लिक गर्नुहुन्छ, पाठ लेख्नुहुन्छ, विकल्प छान्नुहुन्छ आदि।

### कन्सोल र UI एपहरू सीमित छन्

कमाण्ड-आधारित एपसँग तुलना गर्दा:

- **यो सीमित छ**। तपाईंले जुनसुकै कमाण्ड टाइप गर्न सक्नुहुन्न, केवल एपले समर्थन गरेका कमाण्ड मात्र।
- **भाषा-विशेष**। केही एपहरूले धेरै भाषा समर्थन गर्छन्, तर डिफल्टमा एप कुनै एक भाषाका लागि बनाइएको हुन्छ, यद्यपि थप भाषा समर्थन गर्न सकिन्छ।

### पाठ उत्पादन एपका फाइदाहरू

त्यसो भए पाठ उत्पादन एप किन फरक छ?

पाठ उत्पादन एपमा तपाईंलाई बढी स्वतन्त्रता हुन्छ, तपाईं कुनै निश्चित कमाण्ड वा इनपुट भाषामा मात्र सीमित हुनुहुन्न। बरु, तपाईंले प्राकृतिक भाषा प्रयोग गरेर एपसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा के छ भने, तपाईंले पहिले नै ठूलो जानकारीमा तालिम पाएको डाटा स्रोतसँग काम गर्नुहुन्छ, जबकि परम्परागत एपमा डाटाबेसमा भएको कुरामा मात्र सीमित हुनुपर्छ।

### पाठ उत्पादन एपले के बनाउन सकिन्छ?

धेरै कुरा बनाउन सकिन्छ। उदाहरणका लागि:

- **च्याटबोट**। च्याटबोट जसले तपाईंको कम्पनी र उत्पादनहरूबारे प्रश्नहरूको उत्तर दिन्छ, राम्रो विकल्प हुन सक्छ।
- **सहायक**। LLM हरू पाठ संक्षेप गर्न, पाठबाट जानकारी निकाल्न, रिजुमे जस्ता पाठ उत्पादन गर्न आदि काममा उत्कृष्ट छन्।
- **कोड सहायक**। तपाईंले जुन भाषा मोडेल प्रयोग गर्नुहुन्छ त्यसअनुसार, तपाईंले कोड लेख्न सहयोग गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, GitHub Copilot वा ChatGPT जस्ता उत्पादनहरू प्रयोग गर्न सकिन्छ।

## कसरी सुरु गर्ने?

तपाईंले LLM सँग एकीकृत हुने तरिका खोज्नुपर्छ, जुन सामान्यतया दुई तरिकाले हुन्छ:

- API प्रयोग गर्नुहोस्। यहाँ तपाईंले आफ्नो प्रॉम्प्टसहित वेब अनुरोध बनाउनुहुन्छ र उत्पन्न पाठ फिर्ता पाउनुहुन्छ।
- लाइब्रेरी प्रयोग गर्नुहोस्। लाइब्रेरीहरूले API कलहरूलाई सजिलो बनाउँछन् र प्रयोग गर्न सहज बनाउँछन्।

## लाइब्रेरीहरू/SDK हरू

LLM सँग काम गर्नका लागि केही प्रख्यात लाइब्रेरीहरू छन्:

- **openai**, यसले तपाईंको मोडेलमा जडान गर्न र प्रॉम्प्ट पठाउन सजिलो बनाउँछ।

त्यसपछि केही लाइब्रेरीहरू छन् जसले अझ उच्च स्तरमा काम गर्छन्:

- **Langchain**। Langchain प्रख्यात छ र Python समर्थन गर्छ।
- **Semantic Kernel**। Semantic Kernel Microsoft द्वारा बनाइएको लाइब्रेरी हो, जसले C#, Python, र Java समर्थन गर्छ।

## ChatCompletions प्रयोग गरेर पाठ उत्पादन गर्नुहोस्

पाठ उत्पादन गर्ने तरिका भनेको `ChatCompletionsClient` क्लास प्रयोग गर्नु हो।
`samples/python/azure_ai_inference/basic.py` मा, response सेक्सनको कोडमा, user role को content parameter तलको जस्तो परिवर्तन गर्नुहोस्:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट गरिएको फाइल चलाएर नतिजा हेर्नुहोस्


## फरक-फरक कुराहरूका लागि फरक-फरक प्रकारका प्रॉम्प्टहरू

अब तपाईंले प्रॉम्प्ट प्रयोग गरेर कसरी पाठ उत्पन्न गर्ने देख्नुभयो। तपाईंले चलिरहेको एउटा कार्यक्रम पनि पाउनुभएको छ जसलाई तपाईंले परिवर्तन गरेर विभिन्न प्रकारका पाठहरू बनाउन सक्नुहुन्छ।

प्रॉम्प्टहरू धेरै किसिमका कामका लागि प्रयोग गर्न सकिन्छ। उदाहरणका लागि:

- **पाठको कुनै प्रकार उत्पन्न गर्नुहोस्**। जस्तै, तपाईंले कविता, क्विजका लागि प्रश्नहरू आदि बनाउन सक्नुहुन्छ।
- **जानकारी खोज्नुहोस्**। तपाईंले प्रॉम्प्ट प्रयोग गरेर जानकारी खोज्न सक्नुहुन्छ, जस्तै 'वेब विकासमा CORS भन्नाले के जनाउँछ?'।

## अभ्यास - रेसिपी जेनरेटर बनाउनुहोस्

अब हामीले एउटा परिदृश्य खेलेका छौं, अब देखाइएको परिदृश्यसँग मेल खाने कोड लेखौं। यसका लागि, तलका चरणहरू पालना गर्नुहोस्:

1. पहिलेबाट रहेको फाइललाई सुरुवातको आधारको रूपमा प्रयोग गर्नुहोस्
1. एउटा `prompt` भेरिएबल बनाउनुहोस् र नमुना कोडलाई तल देखिएझैं परिवर्तन गर्नुहोस्:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

ठीक छ, अब हेरौं कसरी सुधार गर्न सकिन्छ। सुधार गर्नका लागि, हामी चाहन्छौं कि कोड लचिलो होस्, ताकि सामग्रीहरू र रेसिपीहरूको संख्या सजिलै परिवर्तन गर्न र सुधार गर्न सकियोस्।


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

   टेम्परेचर परिवर्तन गर्न, तपाईंले `temperature` प्यारामिटर प्रयोग गर्न सक्नुहुन्छ। उदाहरणका लागि, यदि तपाईं 0.5 टेम्परेचर प्रयोग गर्न चाहनुहुन्छ भने, यसरी गर्नुहोस्:

```python
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
            },
            {
                "role": "user",
                "content": new_prompt,
            },
        ],
        model=model_name,
        # Optional parameters
        temperature=0.5,
        max_tokens=1200,
        top_p=1.    
    )
```

   > ध्यान दिनुहोस्, १.० नजिक जाँदा नतिजा अझ विविध हुन्छ।


## असाइनमेन्ट

यस असाइनमेन्टका लागि, तपाईंले के बनाउने भन्ने रोज्न सक्नुहुन्छ।

- रेसिपी जेनेरेटर एपलाई अझ सुधार गर्नुहोस्। टेम्परेचरका मानहरू र प्रॉम्प्टहरूमा खेल्दै हेर्नुहोस् के नयाँ आउँछ।
- "स्टडी बडी" बनाउनुहोस्। यो एपले कुनै विषयमा प्रश्नहरूको उत्तर दिन सक्नुपर्छ, जस्तै Python। तपाईंले प्रॉम्प्टहरू राख्न सक्नुहुन्छ, "Python मा फलानो विषय के हो?" वा "फलानो विषयको कोड देखाऊ" आदि।
- इतिहास बोट, इतिहासलाई जीवन्त बनाउनुहोस्, बोटलाई कुनै ऐतिहासिक पात्रको रूपमा प्रस्तुत गर्न लगाउनुहोस् र उसको जीवन र समयका बारेमा प्रश्न सोध्नुहोस्।
