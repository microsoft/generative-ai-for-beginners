# पाठ उत्पादन अनुप्रयोगहरू बनाउ

तपाइँले अहिलेसम्म यस पाठक्रममार्फत देख्नुभएको छ कि प्रम्प्टहरू जस्ता मूल अवधारणाहरू छन् र "प्रम्प्ट इञ्जिनियरिङ" नामक एउटा पूर्ण विषय क्षेत्र पनि छ। धेरै उपकरणहरू जससँग तपाइँ अन्तरक्रिया गर्न सक्नुहुन्छ जस्तै ChatGPT, Office 365, Microsoft Power Platform र अरूले केही accomplish गर्न प्रम्प्टहरूको प्रयोग गर्दै तपाइँलाई समर्थन गर्छन्।

तपाइँले यस्तो अनुभवलाई कुनै अनुप्रयोगमा थप्नको लागि, तपाइँले प्रम्प्टहरू, कम्प्लिशन्सहरू जस्ता अवधारणा बुझ्नुपर्नेछ र काम गर्नको लागि कुनै पुस्तकालय छनौट गर्नुपर्नेछ। ठीक यही कुरा तपाइँ यस अध्यायमा सिक्नुहुनेछ।

## परिचय

यस अध्यायमा, तपाइँ:

- openai पुस्तकालय र यसको मूल अवधारणाहरूको बारेमा जान्नुहोस्।
- openai प्रयोग गरेर पाठ उत्पादन अनुप्रयोग बनाउनुहोस्।
- पाठ उत्पादन अनुप्रयोग बनाउनको लागि प्रम्प्ट, तापक्रम, र टोकन्स जस्ता अवधारणाहरू कसरी प्रयोग गर्ने बुझ्नुहोस्।

## सिकाइ लक्ष्यहरू

यस पाठको अन्त्यमा, तपाइँ सक्षम हुनुहुनेछ:

- पाठ उत्पादन अनुप्रयोग के हो व्याख्या गर्नुहोस्।
- openai प्रयोग गरेर पाठ उत्पादन अनुप्रयोग बनाउनुहोस्।
- अधिक वा कम टोकन्स प्रयोग गर्ने गरी तपाइँको अनुप्रयोगलाई कन्फिगर गर्नुहोस् र साथै विविधतापूर्ण आउटपुटका लागि तापक्रम परिवर्तन गर्नुहोस्।

## पाठ उत्पादन अनुप्रयोग के हो?

सामान्यतया जब तपाइँ कुनै अनुप्रयोग बनाउनुहुन्छ यसले केही प्रकारको इन्टरफेस हुन्छ जस्तै तल:

- कमाण्ड आधारित। कन्सोल अनुप्रयोगहरू सामान्य अनुप्रयोगहरू हुन् जहाँ तपाइँ कमाण्ड टाइप गर्नुहुन्छ र यसले काम पूरा गर्छ। उदाहरणका लागि, `git` कमाण्ड आधारित अनुप्रयोग हो।
- प्रयोगकर्ता इन्टरफेस (UI)। केही अनुप्रयोगहरूमा ग्राफिकल प्रयोगकर्ता इन्टरफेस (GUIs) हुन्छ जहाँ तपाइँ बटनहरु क्लिक गर्नुहुन्छ, पाठ इन्पुट गर्नुहुन्छ, विकल्पहरू चयन गर्नुहुन्छ र थप।

### कन्सोल र UI अनुप्रयोगहरू सीमित छन्

यसलाई कमाण्ड आधारित अनुप्रयोगसँग तुलना गर्नुहोस् जहाँ तपाइँ कमाण्ड टाइप गर्नुहुन्छ:

- **यो सीमित छ**। तपाइँ कुनै पनि कमाण्ड टाइप गर्न सक्नुहुन्न, केवल ती कमाण्डहरू जसलाई अनुप्रयोगले समर्थन गर्छ।
- **भाषा विशेष**। केही अनुप्रयोगहरू धेरै भाषाहरू समर्थन गर्छन्, तर पूर्वनिर्धारित रूपमा अनुप्रयोग विशिष्ट भाषाका लागि बनाइएको हुन्छ, यद्यपि तपाइँ थप भाषा समर्थन थप्न सक्नुहुन्छ।

### पाठ उत्पादन अनुप्रयोगहरूको फाइदा

त्यसैले पाठ उत्पादन अनुप्रयोग कसरी फरक छ?

एउटा पाठ उत्पादन अनुप्रयोगमा, तपाइँसँग बढी लचिलोपन हुन्छ, तपाइँ एउटा सेट कमाण्डहरु वा विशिष्ट इनपुट भाषामा सीमित हुनुहुन्न। यसको सट्टा, तपाइँ प्राकृतिक भाषाको प्रयोग गरेर अनुप्रयोगसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा यो हो कि तपाइँ पहिले नै विशाल जानकारीको कर्पसमा तालिम प्राप्त डाटा स्रोतसँग अन्तरक्रिया गर्दै हुनुहुन्छ, जबकि पारम्परिक अनुप्रयोग डेटाबेसमा के छ भनेर सीमित हुन सक्छ।

### म पाठ उत्पादन अनुप्रयोगसँग के बनाउन सक्छु?

धेरै कुराहरू तपाइँ बनाउन सक्नुहुन्छ। उदाहरणका लागि:

- **एउटा च्याटबोट**। कम्पनी र यसको उत्पादनहरूको विषयमा प्रश्नहरूको उत्तर दिने च्याटबोट राम्रो मेल खान सक्छ।
- **सहायक**। LLM हरू पाठ संक्षेप गर्ने, पाठबाट जानकारी निकाल्ने, पाठ जस्तै रेजुमे बनाउने र अन्य कुराहरूमा उत्कृष्ट छन्।
- **कोड सहायक**। भाषा मोडेल अनुसार, तपाइँ कोड लेख्न मद्दत गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, तपाइँ GitHub Copilot जस्ता उत्पादन र ChatGPT दुवै प्रयोग गरेर कोड लेख्न मद्दत लिन सक्नुहुन्छ।

## म कसरी सुरु गर्न सक्छु?

सायद, तपाइँले LLM सँग інтिग्रेट हुने उपाय खोज्नु पर्छ, जुन प्रायः यी दुई तरिका लिन सक्छ:

- API प्रयोग गर्नुहोस्। यहाँ तपाइँले तपाइँको प्रम्प्टसँग वेब अनुरोधहरू निर्माण गर्नुहुन्छ र तयार गरिएको पाठ प्राप्त गर्नुहुन्छ।
- पुस्तकालय प्रयोग गर्नुहोस्। पुस्तकालयहरूले API कलहरूलाई समेट्छन् र प्रयोग गर्न सजिलो बनाउछन्।

## पुस्तकालयहरू/SDK हरू

LLM सँग काम गर्न लोकप्रिय केही पुस्तकालयहरू छन् जस्तै:

- **openai**, यस पुस्तकालयले तपाइँलाई मोडेलसँग जडान गर्न र प्रम्प्टहरू पठाउन सजिलो बनाउछ।

त्यसपछि उच्च स्तरमा काम गर्ने पुस्तकालयहरू पनि छन् जस्तै:

- **Langchain**। Langchain प्रसिद्ध छ र Python समर्थन गर्दछ।
- **Semantic Kernel**। Semantic Kernel माइक्रोसफ्टको पुस्तकालय हो जसले C#, Python र Java भाषाहरू समर्थन गर्दछ।

## ChatCompletions सँग पाठ उत्पादन गर्नुहोस्

पाठ उत्पादन गर्ने तरिका हो `ChatCompletionsClient` वर्ग प्रयोग गर्नु। 
`samples/python/azure_ai_inference/basic.py` मा, कोडको प्रतिक्रिया भागमा, प्रयोगकर्ता भूमिका अपडेट गर्नुहोस् सामग्री प्यारामिटरलाई तल उल्लेखित अनुसार परिवर्तन गरेर:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # वैकल्पिक प्यारामिटरहरू
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

## अभ्यास - एउटा रेसिपी जेनेरेटर बनाउनुहोस्

अब हामीले एउटा परिदृश्य खेलेका छौं, हामीले देखाइएको परिदृश्यसँग मेल खाने कोड लेखौं। त्यसका लागि, यी चरणहरू पालना गर्नुहोस्:

1. सुरुका लागि अवस्थित फाइल प्रयोग गर्नुहोस्
1. एउटा `prompt` भेरिएबल सिर्जना गर्नुहोस् र नमूना कोड तलको जस्तै परिवर्तन गर्नुहोस्:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. आउनुहोस् तलको तरिकाले कोड परिवर्तन गरौं:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
