# टेक्स्ट जेनेरेशन अनुप्रयोगहरू बनाउनुस्

तपाईंले यस पाठ्यक्रम मार्फत अहिलेसम्म देख्नुभएको छ कि प्रम्प्टहरू जस्ता मूलभूत अवधारणाहरू छन् र "प्रम्प्ट इन्जिनियरिङ" नामक पुरै अनुशासन पनि छ। धेरै उपकरणहरू जससँग तपाईं अन्तरक्रिया गर्न सक्नुहुन्छ, जस्तै ChatGPT, Office 365, Microsoft Power Platform र थप, तपाईंलाई केहि accomplish गर्न प्रम्प्टहरूको प्रयोग गरेर सहयोग गर्छन्।

यस्तो अनुभव एउटा अनुप्रयोगमा थप्नका लागि, तपाईंले प्रम्प्टहरू, कम्प्लिशनहरू जस्ता अवधारणाहरू बुझ्नुपर्छ र काम गर्नको लागि एउटा लाइब्रेरी चयन गर्नुपर्छ। त्यो ठीक के तपाईंले यस अध्यायमा सिक्न लाग्नु भएको छ।

## परिचय

यस अध्यायमा, तपाईंले:

- openai लाइब्रेरी र यसका मूलभूत अवधारणाहरूको बारेमा सिक्नुहुनेछ।
- openai प्रयोग गरेर टेक्स्ट जेनेरेशन एप बनाउनुहोस्।
- प्रम्प्ट, तापक्रम, टोकनहरू जस्ता अवधारणाहरू कसरी प्रयोग गर्ने बुझ्नुहोस् टेक्स्ट जेनेरेशन एप बनाउन।

## सिकाइ लक्ष्यहरू

यो पाठको अन्त्यमा, तपाईं सक्षम हुनुहुनेछ:

- टेक्स्ट जेनेरेशन एप भनेको के हो व्याख्या गर्नुहोस्।
- openai प्रयोग गरेर टेक्स्ट जेनेरेशन एप बनाउनुहोस्।
- तपाईंको एपलाई बढी वा कम टोकनहरू प्रयोग गर्न र तापक्रम परिवर्तन गर्न कन्फिगर गर्नुहोस्, विविध आउटपुटको लागि।

## टेक्स्ट जेनेरेशन एप भनेको के हो?

सामान्यतया, तपाईं अनुप्रयोग बनाउँदा यसमा निम्न जस्तो इन्टरफेस हुन्छ:

- आदेश-आधारित। कन्सोल अनुप्रयोगहरू त्यस्ता अनुप्रयोगहरू हुन् जहाँ तपाईं आदेश टाइप गर्नुहुन्छ र यसले कुनै कार्य पूरा गर्छ। उदाहरणका लागि, `git` आदेश-आधारित एप हो।
- प्रयोगकर्ता इन्टरफेस (UI)। केही अनुप्रयोगहरूमा ग्राफिकल प्रयोगकर्ता इन्टरफेस (GUI) हुन्छ जहाँ तपाईं बटनहरू क्लिक गर्न सक्नुहुन्छ, पाठ इनपुट गर्न सक्नुहुन्छ, विकल्पहरू चयन गर्न सक्नुहुन्छ आदि।

### कन्सोल र UI एपहरू सीमित छन्

आदेश-आधारित एपसँग तुलना गर्दा जहाँ तपाईं आदेश टाइप गर्नुहुन्छ: 

- **यो सीमित छ**। तपाईं कुनै पनि आदेश टाइप गर्न सक्नुहुन्न, केवल ती आदेशहरू जुन एप समर्थन गर्छ।
- **भाषा विशिष्ट**। केही एपहरूले धेरै भाषाहरू समर्थन गर्छन्, तर डिफल्टमा एप विशिष्ट भाषाका लागि बनाइन्छ, यद्यपि तपाईं थप भाषा समर्थन थप्न सक्नुहुन्छ।

### टेक्स्ट जेनेरेशन एपहरूको लाभहरू

टेक्स्ट जेनेरेशन एप कसरी फरक छ?

टेक्स्ट जेनेरेशन एपमा, तपाईंका लागि धेरै लचकता हुन्छ, तपाईं आदेशहरूको सेट वा विशेष इनपुट भाषामा सीमित हुनुहुन्न। यसको सट्टा, तपाईं एपसँग प्राकृतिक भाषा प्रयोग गरी अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को लाभ भनेको तपाईं पहिले नै डेटा स्रोतसँग अन्तरक्रिया गर्दै हुनुहुन्छ जुन ठूलो मात्रामा जानकारीमा प्रशिक्षित छ, जहाँ पारम्परिक एपले सम्भवतः केवल डेटाबेसमा के छ त्यति मात्र सीमित हुन्छ।

### म के के बनाउन सक्छु टेक्स्ट जेनेरेशन एपसँग?

धेरै कुराहरू बनाउन सक्नुहुन्छ। उदाहरणका लागि:

- **च्याटबोट**। कुनै विषयका प्रश्नहरूको उत्तर दिने च्याटबोट, जस्तै तपाईँको कम्पनी र यसको उत्पादनहरूका बारेमा, राम्रो मेल हुन सक्छ।
- **सहायक**। LLMहरूले पाठ सारांश गर्न, पाठबाट अन्तर्दृष्टि प्राप्त गर्न, रिजुमे जस्ता पाठ उत्पादन गर्न उत्कृष्ट छन्।
- **कोड सहायक**। तपाईंले प्रयोग गर्ने भाषाको मोडेलमा भर पर्छ, तपाईं कोड लेख्न मद्दत गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, तपाईं GitHub Copilot र ChatGPT जस्ता उत्पादनहरू प्रयोग गर्न सक्नुहुन्छ।

## म कसरी सुरु गर्न सक्छु?

ठीक छ, तपाईंलाई एउटा LLM सँग एकीकरण गर्ने तरिका खोज्नुपर्छ जुन प्रायः निम्न दुई तरिकाहरू समावेश गर्दछ:

- API प्रयोग गर्नुहोस्। यसमा तपाईं आफ्नो प्रम्प्ट सहित वेब अनुरोधहरू निर्माण गर्दै आउटपुट टेक्स्ट प्राप्त गर्नुहुन्छ।
- लाइब्रेरी प्रयोग गर्नुहोस्। लाइब्रेरीहरूले API कलहरूलाई समावेश गरी सजिलो बनाउँछन्।

## लाइब्रेरीहरू/SDK हरू

LLM सँग काम गर्न केही प्रख्यात लाइब्रेरीहरू छन्:

- **openai**, यो लाइब्रेरीले तपाईंको मोडेलसँग सजिलै जडान गर्न र प्रम्प्टहरू पठाउन मद्दत गर्छ।

त्यसपछि उच्च स्तरमा काम गर्ने लाइब्रेरीहरू पनि छन् जस्तै:

- **Langchain**। Langchain प्रख्यात छ र Python सपोर्ट गर्छ।
- **Semantic Kernel**। Semantic Kernel माइक्रोसफ्टको लाइब्रेरी हो जसले C#, Python, र Java भाषा समर्थन गर्छ।

## ChatCompletions प्रयोग गरेर टेक्स्ट जेनेरेट गर्नुहोस्

टेक्स्ट जेनेरेट गर्नको लागि `ChatCompletionsClient` क्लास प्रयोग गर्ने तरिका हो। 
`samples/python/azure_ai_inference/basic.py` मा, प्रतिक्रिया खण्डको कोडमा, प्रयोगकर्ता भूमिका (user role) को कोड अपडेट गरेर सामग्री प्यारामिटरलाई तल दिइएको अनुसार परिवर्तन गर्नुहोस्:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # वैकल्पिक प्यारामिटरहरू
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट गरिएको फाइल चलाउनुहोस् नतिजा हेर्न

## अभ्यास - एउटा रेसिपी निर्माणकर्ता बनाउनुहोस्

अब जब हामीले एउटा परिदृश्य खेलिसकेका छौं, आउनुस् प्रस्तुत परिदृश्यसँग मेल खाने कोड लेखौं। त्यसका लागि, यी कदमहरू पालना गर्नुहोस्:

1. अवस्थित फाइललाई सुरुवाती बिन्दुको रूपमा प्रयोग गर्नुहोस्
1. एउटा `prompt` भेरिएबल बनाउनुहोस् र नमूना कोडलाई तलको रूपमा परिवर्तन गर्नुहोस्:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

राम्रो, अब हेर्नेछौँ हामी कसरी सुधार गर्न सक्छौं। सुधार गर्न, हामी चाहन्छौं कोड लचिलो होस्, ताकि सामग्रीहरू र परिकारहरूको संख्या सुधार र परिवर्तन गर्न सकिन्छ।


1. हामीले तल दिइएको तरिकाले कोड परिवर्तन गरौं:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

तल केही सुझावहरू छन्:

- रेसिपी जनरेटर एपलाई अझ सुधार गर्न ट्वीक गर्नुहोस्। तातोपन मानहरूको साथ खेल्नुहोस् र प्रॉम्प्टहरू पाल्टाउनुहोस् र हेर्नुहोस् के आउँछ।
- एउटा "अध्ययन साथी" बनाउनुहोस्। यो एपले कुनै विषय जस्तै Python सम्बन्धी प्रश्नहरूको उत्तर दिन सक्षम हुनुपर्छ, जस्तै तपाईंले प्रॉम्प्टहरू यसरी राख्न सक्नुहुन्छ, "Python मा के हो एउटा विषय?", वा "कुनै विषयको कोड देखाऊ" आदि।
- इतिहास बोट, इतिहासलाई जीवित बनाउनुहोस्, बोटलाई कुनै ऐतिहासिक पात्रको भूमिका खेल्न निर्देशन दिनुहोस् र यसको जीवन र समयका बारेमा प्रश्नहरू गर्नुहोस्।
