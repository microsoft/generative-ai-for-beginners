# टेक्स्ट उत्पादन एपहरू बनाउनुहोस्

तपाईंले अहिलेसम्म यो पाठ्यक्रम मार्फत देख्नुभयो कि त्यहाँ मुख्य अवधारणाहरू हुन जस्तै प्रॉम्प्टहरू र यहाँसम्म कि "प्रॉम्प्ट इन्जिनियरिङ" भनिने एक पुरै विषयशास्त्र पनि छ। धेरै उपकरणहरू जससँग तपाईं अन्तरक्रिया गर्न सक्नुहुन्छ जस्तै ChatGPT, Office 365, Microsoft Power Platform र थपले तपाईंलाई केहि पूरा गर्न प्रॉम्प्टहरू प्रयोग गर्न समर्थन गर्छन्।

तपाईंले यस्तो अनुभव कुनै एपमा थप्नको लागि, तपाईंले प्रॉम्प्ट, कम्प्लिशन्स जस्ता अवधारणाहरू बुझ्न आवश्यक छ र काम गर्न एक पुस्तकालय छनोट गर्नुपर्छ। यही कुरा तपाईं यस अध्यायमा सिक्नेछौं।

## परिचय

यस अध्यायमा तपाईंले:

- openai पुस्तकालय र यसको मुख्य अवधारणाहरूबारे सिक्नुहुनेछ।
- openai प्रयोग गरेर टेक्स्ट उत्पादन एप बनाउनुहोस्।
- प्रॉम्प्ट, तापक्रम, र टोकनजस्ता अवधारणाहरू कसरी प्रयोग गर्ने बुझ्नुहोस् टेक्स्ट उत्पादन एप बनाउन।

## सिकाइका लक्ष्यहरू

यस पाठको अन्त्यमा, तपाईं सक्षम हुनुहुनेछ:

- टेक्स्ट उत्पादन एप के हो व्याख्या गर्न।
- openai प्रयोग गरेर टेक्स्ट उत्पादन एप बनाउन।
- आफ्नो एपलाई धेरै वा कम टोकन प्रयोग गर्न र तापक्रम पनि परिवर्तन गर्न कन्फिगर गर्न, विभिन्न आउटपुटका लागि।

## टेक्स्ट उत्पादन एप के हो?

सामान्यतया जब तपाईं एप बनाउनुहुन्छ, त्यसमा कुनै न कुनै इन्टरफेस हुन्छ जस्तै तलको:

- कमाण्ड आधारित। कन्सोल एपहरू सामान्य एपहरू हुन् जहाँ तपाईं कमाण्ड टाइप गर्नुहुन्छ र यसले काम सम्पन्न गर्छ। उदाहरणका लागि, `git` कमाण्ड आधारित एप हो।
- प्रयोगकर्ता इन्टरफेस (UI)। केही एपहरूसँग ग्राफिकल प्रयोगकर्ता इन्टरफेस (GUIs) हुन्छ जहाँ तपाईं बटन क्लिक गर्नुहुन्छ, पाठ इनपुट गर्नुहुन्छ, विकल्प चयन गर्नुहुन्छ र थप।

### कन्सोल र UI एपहरू सीमित छन्

कमाण्ड टाइप गर्ने कमाण्ड-आधारित एपसँग तुलना गर्नुहोस्:

- **यो सीमित छ**। तपाईं जुनसुकै कमाण्ड टाइप गर्न सक्नुहुन्न, केवल एपले समर्थन गर्ने कमाण्डहरू मात्र।
- **भाषा विशिष्ट**। केही एपहरूले धेरै भाषाहरू समर्थन गर्छन्, तर पूर्वनिर्धारित रूपमा एप कुनै विशिष्ट भाषाका लागि बनाइएको हुन्छ, यद्यपि तपाईं थप भाषा समर्थन थप्न सक्नुहुन्छ।

### टेक्स्ट उत्पादन एपका फाइदाहरू

त्यसैले टेक्स्ट उत्पादन एप कसरी फरक हुन्छ?

टेक्स्ट उत्पादन एपमा, तपाईंलाई थप लचकता हुन्छ, तपाईं कमाण्डहरूको सेट अथवा कुनै विशिष्ट इनपुट भाषामा सीमित हुनुहुन्न। त्यसको सट्टा, तपाईं प्राकृतिक भाषाको प्रयोग गरेर एपसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा भनेको तपाईं पहिल्यै डाटा स्रोतसँग अन्तरक्रिया गर्दै हुनुहुन्छ जुन विशाल सूचना संग्रहमा प्रशिक्षित छ, जहाँ परम्परागत एपहरू डाटाबेसमा के छ त्यसमा सीमित हुन सक्छन्।

### टेक्स्ट उत्पादन एपबाट म के बनाउँ?

तपाईं धेरै कुराहरू बनाउन सक्नुहुन्छ। उदाहरणका लागि:

- **एक चैटबोट**। तपाईँको कम्पनी र यसको उत्पादनहरूको बारेमा प्रश्नहरूको जवाफ दिने एक चैटबोट राम्रो मेल हुन सक्छ।
- **सहायक**। LLMहरूले जस्तै पाठ संक्षेप गर्ने, पाठबाट अन्तर्दृष्टि प्राप्त गर्ने, रिजुमे जस्ता पाठ उत्पादन गर्ने कामहरूमा उत्कृष्ट छन्।
- **कोड सहायक**। तपाईं प्रयोग गर्ने भाषा मोडेल अनुसार, तपाईं कोड लेख्न मद्दत गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, GitHub Copilot जस्ता उत्पादनहरू र ChatGPT दुवै प्रयोग गरेर तपाईं कोड लेख्न सहयोग गर्न सक्नुहुन्छ।

- API प्रयोग गर्नुहोस्। यहाँ तपाईं आफ्नो प्रॉम्प्टसँग वेब अनुरोधहरू निर्माण गर्नुहुन्छ र परिणामस्वरूप उत्पन्न टेक्स्ट प्राप्त गर्नुहुन्छ।
- पुस्तकालय प्रयोग गर्नुहोस्। पुस्तकालयहरूले API कलहरूलाई समेट्छ र प्रयोग गर्न सजिलो बनाउँछन्।

## पुस्तकालयहरू/SDKs

LLM सँग काम गर्न केहि परिचित पुस्तकालयहरू छन् जस्तै:

### OpenAI कन्फिगरेसन सेटअप गर्नुहोस्

client = OpenAI(

माथि हामीले तलका कुरा सेट गर्दैछौं:

> [!NOTE]
> `os.environ` एउटा फन्क्सन हो जुन वातावरण चलकहरू पढ्छ। तपाईं यसको प्रयोग गरेर `OPENAI_API_KEY` जस्ता वातावरण चलकहरू पढ्न सक्नुहुन्छ।

## टेक्स्ट उत्पादन

टेक्स्ट उत्पादन गर्ने तरिका भनेको `responses.create` विधि प्रयोग गर्नु हो। यहाँ एक उदाहरण छ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

माथिको कोडमा, हामी एउटा प्रतिक्रिया वस्तु सिर्जना गर्छौं र प्रयोग गर्न चाहिने मोडेल र प्रॉम्प्ट पास गर्छौं। त्यसपछि हामी उत्पन्न गरिएको टेक्स्ट प्रिन्ट गर्छौं।

### चैट प्रतिक्रियाहरू

अहिलेसम्म, तपाईंले देख्नुभयो कि हामी कसरी एउटा प्रॉम्प्टबाट टेक्स्ट उत्पादन गर्न सक्छौं। Responses API चैटबोटहरूको लागि पनि उपयुक्त छ - तपाईं `input` मा सन्देशहरूको सूची प्रदान गर्नुहुन्छ जुन संवाद निर्माण गर्छ। यहाँ यसको प्रयोगको उदाहरण छ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

यस कार्यक्षमतामा थप जानकारी आगामी अध्यायमा हुनेछ।

## अभ्यास - तपाइँको पहिलो टेक्स्ट उत्पादन एप

  > [!NOTE] यदि तपाईं यो नोटबुक Codespaces मा वा Devcontainer भित्र चलाउनुहुन्छ भने यो कदम आवश्यक छैन

> [!NOTE]
> तपाईंको OpenAI कुञ्जी पत्ता लगाउन https://platform.openai.com/settings/organization/api-keys मा जानुहोस् र `API keys` खोज्नुहोस्। तपाईं त्यहाँ नयाँ कुञ्जी सिर्जना गर्न सक्नुहुन्छ र तुरुन्तै मान प्रतिलिपि गर्न सक्नुहुन्छ।


1. एक *app.py* फाइल बनाउनुहोस् र यसमा तलको कोड राख्नुहोस्:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    तपाईले तलको जस्तो आउटपुट देख्नु पर्नेछ:

    ```output

    एक पटकको कुरा हो एक धेरै दुखी जलपरि थिइन्।
    ```


## फरक-फरक कुराहरूको लागि फरक-फरक प्रकारका प्रॉम्प्टहरू  

अहिले तपाईंले देख्नुभएको छ कि कसरी प्रॉम्प्ट प्रयोग गरेर टेक्स्ट जनरेट गर्ने। तपाईं सँग एउटा चलिरहेको प्रोग्राम पनि छ जुन तपाईंले फरक-फरक प्रकारको टेक्स्ट बनाउन परिमार्जन र परिवर्तन गर्न सक्नुहुन्छ।  

प्रॉम्प्टहरू सबै प्रकारका कामहरूका लागि प्रयोग गर्न सकिन्छ। जस्तै:  

- **कुनै प्रकारको टेक्स्ट जनरेट गर्नुहोस्**। उदाहरणका लागि, तपाईं कविताहरु, क्विजका प्रश्नहरू आदि तयार पार्न सक्नुहुन्छ।  
- **जानकारी खोज्नुहोस्**। तपाईं प्रॉम्प्टहरू प्रयोग गरेर यस प्रकारको जानकारी खोज्न सक्नुहुन्छ: ‘वेब विकासमा CORS को के अर्थ हो?’  
- **कोड जनरेट गर्नुहोस्**। तपाईं प्रॉम्प्टहरू प्रयोग गरेर कोड तयार पार्न सक्नुहुन्छ, जस्तै इमेललाई प्रमाणित गर्न रेजुलर एक्सप्रेशन विकास गर्ने, वा पुरा वेब एप जस्ता प्रोग्राम बनाउन पनि सक्नुहुन्छ।

## अभ्यास - एक रेसिपी जनरेटर बनाउनुहोस्

अब जब हामीले एउटा परिदृश्य खेलिसकेका छौं, आउनुहोस् कोड लेखौं जुन प्रदर्शन गरिएको परिदृश्यसँग मेल खान्छ। यसका लागि, तलका चरणहरू पालना गर्नुहोस्:

1. सुरु गर्नका लागि अवस्थित *app.py* फाइल प्रयोग गर्नुहोस्
1. `prompt` भेरिएबल खोज्नुहोस् र यसको कोड तलको अनुसार परिवर्तन गर्नुहोस्:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

हामीसँग अहिले एउटा कार्यरत एप छ जसले रेसिपीहरू उत्पादन गर्न सक्छ र यो लचिलो छ किनभने यो प्रयोगकर्ताबाट इनपुटहरूमा निर्भर गर्दछ, रेसिपीहरूको संख्या मात्र होइन तर प्रयोग गरिएका सामग्रीहरूमा पनि।

   तापक्रम बदल्न, तपाईं `temperature` प्यारामीटर प्रयोग गर्न सक्नुहुन्छ। उदाहरणका लागि यदि तपाईं 0.5 तापक्रम चाहनुहुन्छ भने यसरी गर्नुहुन्छ:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```
