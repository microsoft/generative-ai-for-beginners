# पाठ उत्पन्न गर्ने अनुप्रयोगहरू बनाउनुहोस्

तपाईंले अहिलेसम्मको यस पाठ्यक्रममा देख्नुभएको छ कि प्रॉम्प्टहरू जस्ता मुख्य अवधारणाहरू छन् र यहाँ "प्रॉम्प्ट इन्जिनियरिङ" भनिने एक पूरा विषयशास्त्र पनि छ। तपाईंले अन्तरक्रिया गर्न सक्ने धेरै उपकरणहरू जस्तै ChatGPT, Office 365, Microsoft Power Platform र अन्यले तपाईंलाई केहि के गर्ने कार्य पुरा गर्न प्रॉम्प्टहरू प्रयोग गर्न मद्दत गर्छन्।

तपाईंले यस्तो अनुभवलाई अनुप्रयोगमा थप्नको लागि प्रॉम्प्टहरू, पूरा गर्ने वस्तुहरू जस्ता अवधारणाहरू बुझ्नुपर्नेछ र काम गर्न एउटा पुस्तकालय छान्नुपर्नेछ। ठीक त्यही कुरा तपाईं यस अध्यायमा सिक्नेछौं।

## परिचय

यस अध्यायमा, तपाईंले:

- openai पुस्तकालय र यसको मुख्य अवधारणाहरूको बारेमा सिक्नुहुनेछ।
- openai प्रयोग गरेर पाठ उत्पन्न गर्ने अनुप्रयोग बनाउनुहोस्।
- प्रॉम्प्ट, तापमान, र टोकन जस्ता अवधारणाहरू कसरी प्रयोग गर्ने बुझ्नुहोस् ताकि पाठ उत्पन्न गर्ने अनुप्रयोग बनाउन सक्नुहोस्।

## सिकाइ लक्ष्यहरू

यस पाठको अन्त्यमा, तपाईं सक्षम हुनुहुनेछ:

- पाठ उत्पन्न गर्ने अनुप्रयोग के हो भन्ने व्याख्या गर्न।
- openai प्रयोग गरेर पाठ उत्पन्न गर्ने अनुप्रयोग बनाउनुहोस्।
- तपाईंको अनुप्रयोगलाई थप वा कम टोकनहरू प्रयोग गर्न र तापमान परिवर्तन गर्न कन्फिगर गर्न सक्नुहुन्छ, फरक प्रकारको परिणाम प्राप्त गर्न।

## पाठ उत्पन्न गर्ने अनुप्रयोग के हो?

सामान्यतया जब तपाईं अनुप्रयोग बनाउनुहुन्छ, यसमा केही प्रकारको इन्टरफेस हुन्छ जस्तै:

- कमाण्ड-आधारित। कन्सोल अनुप्रयोगहरू साधारण अनुप्रयोगहरू हुन् जहाँ तपाईं कमाण्ड टाइप गर्नुहुन्छ र यो कुनै कार्य पूरा गर्छ। उदाहरणका लागि, `git` कमाण्ड-आधारित अनुप्रयोग हो।
- प्रयोगकर्ता इन्टरफेस (UI)। केहि अनुप्रयोगहरूमा ग्राफिकल प्रयोगकर्ता इन्टरफेस (GUI) हुन्छ जहाँ तपाईं बटन क्लिक गर्नुहुन्छ, पाठ इनपुट गर्नुहुन्छ, विकल्पहरू छान्नुहुन्छ र अन्य।

### कन्सोल र UI अनुप्रयोगहरू सीमित छन्

एउटा कमाण्ड-आधारित अनुप्रयोगसँग तुलना गर्नुहोस् जहाँ तपाईं कमाण्ड टाइप गर्नुहुन्छ:

- **यो सीमित छ**। तपाईं कुनै पनि कमाण्ड टाइप गर्न सक्नुहुन्न, केवल ती कमाण्डहरू जुन अनुप्रयोगले समर्थन गर्छ।
- **भाषा-विशेष**। केही अनुप्रयोगहरू धेरै भाषाहरू समर्थन गर्छन्, तर डिफल्टमा अनुप्रयोग विशेष भाषा लागि बनाइएको हुन्छ, यद्यपि तपाईं थप भाषा समर्थन थप्न सक्नुहुन्छ।

### पाठ उत्पन्न गर्ने अनुप्रयोगहरूको फाइदाहरू

त्यसैले पाठ उत्पन्न गर्ने अनुप्रयोग कसरी फरक छ?

पाठ उत्पन्न गर्ने अनुप्रयोगमा, तपाईंलाई बढी लचकता हुन्छ, तपाईं कमाण्डहरूको सेट वा विशेष इनपुट भाषामा सीमित हुनुहुन्न। यसको सट्टामा, तपाईं प्राकृतिक भाषा प्रयोग गरेर अनुप्रयोगसँग अन्तरक्रिया गर्न सक्नुहुन्छ। अर्को फाइदा भनेको तपाईं पहिले नै यति धेरै सूचनामाथि प्रशिक्षण गरिएको डाटा स्रोतसँग अन्तरक्रिया गर्दै हुनुहुन्छ, जबकि पारम्परिक अनुप्रयोगमा डेटाबेसमा के छ भन्ने सीमित हुन सक्छ।

### म के निर्माण गर्न सक्छु पाठ उत्पन्न गर्ने अनुप्रयोगसँग?

धेरै कुरा निर्माण गर्न सकिन्छ। उदाहरणका लागि:

- **एक च्याटबट**। कम्पनी र यसको उत्पादनहरू जस्ता विषयहरूमा प्रश्नको उत्तर दिने च्याटबट राम्रो छन।
- **सहायक**। LLMहरू पाठ संक्षेप गर्दा, पाठबाट जानकारी निकाल्दा, रिज्युमे जस्ता पाठ उत्पादन गर्दा आदि उत्कृष्ट हुन्छन्।
- **कोड सहायक**। तपाईंले प्रयोग गर्ने भाषा मोडेल अनुसार, तपाईं कोड लेख्न मद्दत गर्ने कोड सहायक बनाउन सक्नुहुन्छ। उदाहरणका लागि, GitHub Copilot र ChatGPT दुवैलाई कोड लेख्न मद्दत गर्न प्रयोग गर्न सकिन्छ।

## कसरी सुरु गर्ने?

तपाईले प्रायः दुई तरिकाले LLMसँग इन्टिग्रेट गर्नुपर्ने हुन्छ:

- API प्रयोग गर्नुहोस्। यहाँ तपाईं प्रॉम्प्टसहित वेब अनुरोधहरु बनाउनुहुन्छ र उत्पन्न पाठ प्राप्त गर्नुहुन्छ।
- पुस्तकालय प्रयोग गर्नुहोस्। पुस्तकालयहरूले API कललाई समेट्छन् र प्रयोग गर्न सजिलो बनाउँछन्।

## पुस्तकालयहरू/SDKहरू

LLMसँग काम गर्ने केही प्रख्यात पुस्तकालयहरू छन् जस्तै:

- **openai**, यो पुस्तकालयले तपाईंको मोडेलसँग जडान गर्न र प्रॉम्प्ट पठाउन सजिलो बनाउँछ।

त्यसपछि उच्च स्तरमा सञ्चालन गर्ने पुस्तकालयहरू छन् जस्तै:

- **Langchain**। Langchain प्रख्यात छ र Python समर्थन गर्छ।
- **Semantic Kernel**। Semantic Kernel Microsoft को पुस्तकालय हो जसले C#, Python, र Java भाषाहरूलाई समर्थन गर्छ।

### OpenAI कन्फिगरेसन सेटअप गर्नुहोस्

यदि तपाईं OpenAI प्रयोग गर्दै हुनुहुन्छ भने, यो रमाइला कन्फिगरेसन सेटअप गर्ने तरिका हो:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

माथि हामीले तलका कुरा सेट गर्दैछौं:

- `api_key`, यो तपाईंको OpenAI ड्यासबोर्डमा भेटिने API कुञ्जी हो।
- `deployment`, यो तपाईंको GPT संस्करण हो।

> [!NOTE]
> `os.environ` वातावरण चरहरू पढ्ने एक function हो। तपाईं यसलाई `OPENAI_API_KEY` जस्ता वातावरण चरहरू पढ्न प्रयोग गर्न सक्नुहुन्छ।

## पाठ उत्पन्न गर्नुहोस्

पाठ उत्पन्न गर्ने तरिका भनेको `responses.create` विधि प्रयोग गर्नु हो। यहाँ एउटा उदाहरण छ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

माथिको कोडमा, हामी एउटा प्रतिक्रिया वस्तु सिर्जना गर्छौं र प्रयोग गर्ने मोडेल र प्रॉम्प्ट पास गर्छौं। त्यसपछि उत्पन्न भएको पाठ प्रिन्ट गर्छौं।

### च्याट प्रतिक्रियाहरू

अहिलेसम्म, तपाईंले एउटा प्रॉम्प्टबाट कसरी पाठ उत्पन्न गर्ने देख्नुभयो। Responses API च्याटबटहरूको लागि पनि अति उपयुक्त छ - तपाईं सन्देशहरूको सूची `input` को रूपमा प्रदान गरेर संवाद बनाउनुहुन्छ। यहाँ यसको प्रयोगको एक उदाहरण छ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

यस कार्यक्षमतामा अर्को अध्यायमा थप जानकारी हुनेछ।

## अभ्यास - तपाईंको पहिलो पाठ उत्पन्न गर्ने अनुप्रयोग

अब हामीले OpenAI सेवा कसरी सेट अप र कन्फिगर गर्ने सिकिसकेका छौं, आफ्नो पहिलो पाठ उत्पन्न गर्ने अनुप्रयोग बनाउने बेला आएको छ। आफ्नो अनुप्रयोग बनाउनका लागि, यी चरणहरू पालना गर्नुहोस्:


1. एक भर्चुअल वातावरण सिर्जना गर्नुहोस् र openai स्थापना गर्नुहोस्:

  > [!NOTE] यदि तपाईंले यो नोटबुक Codespaces मा वा Devcontainer भित्र चलाउनुहुन्छ भने यो चरण आवश्यक छैन


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> यदि तपाईं Windows प्रयोग गर्दै हुनुहुन्छ भने `source venv/bin/activate` को सट्टा `venv\Scripts\activate` टाइप गर्नुहोस्। 

> [!NOTE]
> आफ्नो OpenAI कुञ्जी फेला पार्न https://platform.openai.com/settings/organization/api-keys मा जानुहोस् र `API keys` खोज्नुहोस्। त्यहाँ तपाईं नयाँ कुञ्जी सिर्जना गर्न सक्नुहुन्छ र तुरुन्तै मान प्रतिलिपि गर्न सक्नुहुन्छ।


1. *app.py* नामक फाइल बनाउनुहोस् र यसलाई तलको कोड दिनुहोस्:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    तपाईंले निम्न जस्तो आउटपुट देख्नुपर्छ:

    ```output
     धेरै दुःखी _____.

    एक पटकको कुरा हो, एक धेरै दुःखी जलपरि थिइन्।
    ```


## फरक कुराहरूका लागि फरक प्रकारका प्रॉम्प्टहरू

अब तपाईंले प्रॉम्प्ट प्रयोग गरेर कसरी टेक्स्ट सिर्जना गर्ने देख्नुभयो। तपाईं सँग एउटा कार्यक्रम पनि चलिरहेको छ जुन तपाईंले फरक प्रकारको टेक्स्ट सिर्जना गर्न परिमार्जन र परिवर्तन गर्न सक्नुहुन्छ। 

प्रॉम्प्टहरू सबै प्रकारका कार्यहरूका लागि प्रयोग गर्न सकिन्छन्। उदाहरणका लागि:

- **टिप्पणी गरेको प्रकारको टेक्स्ट सिर्जना गर्नुहोस्**। उदाहरणका लागि, तपाईं कविता, क्विजको प्रश्नहरू आदि सिर्जना गर्न सक्नुहुन्छ।
- **जानकारी खोज्नुहोस्**। तपाईंले प्रॉम्प्टहरू प्रयोग गरेर तलको उदाहरण जस्तै 'वेब विकासमा CORS को अर्थ के हो?' भन्ने जानकारी खोज्न सक्नुहुन्छ।
- **कोड सिर्जना गर्नुहोस्**। तपाईंले प्रॉम्प्टहरू प्रयोग गरेर कोड सिर्जना गर्न सक्नुहुन्छ, उदाहरणका लागि इमेल प्रमाणीकरण गर्न रजुलर एक्सप्रेशन विकास गर्ने वा पुरै प्रोग्राम जस्तै वेब एप पनि सिर्जना गर्न सक्नुहुन्छ।

## अभ्यास - एक रेसिपी जेनेरेटर बनाउनुहोस्

अब जब हामीले एक परिदृश्य खेलिसकेका छौं, आउनुहोस् प्रदर्शन गरिएको परिदृश्यसँग मेल खाने कोड लेखौं। त्यसका लागि, यी चरणहरू पालन गर्नुहोस्:

1. सुरु गर्नका लागि अवस्थित *app.py* फाइल प्रयोग गर्नुहोस्
1. `prompt` भेरिएबल खोज्नुहोस् र यसको कोडलाई निम्नमा परिवर्तन गर्नुहोस्:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

राम्रो, अब हामी कसरी कुरा सुधार्न सक्छौं हेर्नुहोस्। कुरा सुधार्नका लागि, हामी कोड लचिलो बनाउने सुनिश्चित गर्न चाहन्छौं, ताकि सामग्रीहरू र पकाउने विधिहरूको संख्या सुधार गर्न र परिवर्तन गर्न सकियोस्।

यस कार्यका लागि, तपाईं के निर्माण गर्न चाहानुहुन्छ चयन गर्न सक्नुहुन्छ।

यहाँ केही सुझावहरू छन्:

- रेसिपी जनरेटर एपलाई अझ सुधार गर्न टिंकर गर्नुहोस्। तापमान मानहरूसँग खेल्नुहोस् र प्रॉम्प्टहरू परीक्षण गर्नुहोस् के तपाईंले के पाउन सक्नुहुन्छ भनेर हेर्नुहोस्।
- "स्टडी बडी" बनाउनुहोस्। यो एपले कुनै विषयवस्तु सम्बन्धी प्रश्नहरूको उत्तर दिन सक्छ, उदाहरणका लागि Python, तपाईंमा यस्तो प्रॉम्प्ट हुन सक्छ "Python को कुनै विषय थाहा छ?" वा "कुनै विषयको कोड देखाउ" आदि।
- इतिहास बोट, इतिहासलाई जिवन्त बनाउन, बोटलाई कुनै ऐतिहासिक पात्रको भूमिका निभाउन निर्देशन दिनुहोस् र उसका जीवन र समयका बारेमा प्रश्नहरू सोध्नुहोस्।
