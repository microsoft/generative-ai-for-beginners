# प्रॉम्प्ट इन्जिनियरिङको परिचय
प्रॉम्प्ट इन्जिनियरिङ प्राकृतिक भाषा प्रसोधन कार्यहरूको लागि प्रॉम्प्टहरू डिजाइन र अनुकूलन गर्ने प्रक्रिया हो। यसले सही प्रॉम्प्टहरू चयन गर्ने, तिनीहरूको प्यारामिटरहरू ट्युन गर्ने, र तिनीहरूको प्रदर्शन मूल्यांकन गर्ने समावेश गर्दछ। प्रॉम्प्ट इन्जिनियरिङ उच्च शुद्धता र दक्षता प्राप्त गर्नका लागि अत्यन्त महत्त्वपूर्ण छ। यस भागमा, हामी OpenAI मोडेलहरू प्रयोग गरेर प्रॉम्प्ट इन्जिनियरिङका आधारभूत कुरा अन्वेषण गर्नेछौं।


### अभ्यास १: टोकनाइजेशन
OpenAI बाट खुला स्रोतको छिटो टोकनाइजर tiktoken प्रयोग गरेर टोकनाइजेशन अन्वेषण गर्नुहोस्
थप उदाहरणहरूको लागि [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) हेर्नुहोस्।


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### अभ्यास २: OpenAI API कुञ्जी सेटअप प्रमाणीकरण गर्नुहोस्

तपाइँको OpenAI अन्त्यबिन्दु सही रूपमा सेटअप गरिएको छ कि छैन भनी जाँच गर्न तलको कोड चलाउनुहोस्। कोडले केवल एक सरल आधारभूत प्रॉम्प्ट प्रयास गर्छ र सम्पन्नता प्रमाणीकरण गर्छ। इनपुट `oh say can you see` को पूर्णता `by the dawn's early light..` को वरिपरि हुनुपर्छ।


````python
# Uses the OpenAI client with the Responses API.
# See https://platform.openai.com/docs/api-reference/responses

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### अभ्यास ३: बनाईएका कुरा
यस्तो विषयमा पूर्णता फिर्ता गर्न LLM लाई सोध्दा के हुन्छ जाँच गर्नुहोस् जुन विषय हुनसक्न सक्छ वा त्यो विषयमा LLM लाई ज्ञान नहुन सक्छ किनकि त्यो यसको पूर्व-प्रशिक्षित डेटासेट बाहिरका (अझैं नयाँ) विषयहरू हुनसक्छन्। जब तपाईं अर्को प्रेरणा वा विभिन्न मोडेल प्रयास गर्नुहुन्छ तब प्रतिक्रिया कसरी परिवर्तन हुन्छ हेर्नुहोस्।


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### अभ्यास ४: निर्देशन आधारित 
मुख्य सामग्री सेट गर्न "text" भेरिएबल प्रयोग गर्नुहोस् 
र त्यस मुख्य सामग्रीसँग सम्बन्धित निर्देशन दिन "prompt" भेरिएबल प्रयोग गर्नुहोस्।

यहाँ हामी मोडेललाई दोस्रो कक्षा विद्यार्थीका लागि पाठलाई संक्षेप गर्न अनुरोध गर्छौं


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### अभ्यास ५: जटिल संकेत 
प्रणाली, प्रयोगकर्ता र सहायक सन्देशहरू भएको अनुरोध प्रयास गर्नुहोस् 
प्रणालीले सहायक सन्दर्भ सेट गर्दछ
प्रयोगकर्ता र सहायक सन्देशहरूले बहु-पटक संवाद सन्दर्भ प्रदान गर्दछन्

प्रणाली सन्दर्भमा सहायक व्यक्तित्व "व्यंग्यात्मक" कसरी सेट गरिएको छ ध्यान दिनुहोस्। 
फरक प्रकारको व्यक्तित्व सन्दर्भ प्रयोग गरेर हेर्नुहोस्। वा फरक श्रृंखला इनपुट/आउटपुट सन्देशहरूको प्रयास गर्नुहोस्


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
