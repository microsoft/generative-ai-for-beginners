## DALL-E र Midjourney के हुन्?

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) र [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) दुई सबैभन्दा लोकप्रिय छवि उत्पादन मोडेलहरू हुन्, जसले तपाईंलाई prompt प्रयोग गरेर छवि उत्पादन गर्न दिन्छन्।

### DALL-E

DALL-E बाट सुरु गरौं, जुन एक Generative AI मोडेल हो जसले पाठ विवरणहरूबाट छवि उत्पादन गर्छ।

> [DALL-E दुई मोडेलको संयोजन हो, CLIP र diffused attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)।

- **CLIP** एउटा मोडेल हो जसले छवि र पाठबाट embedding (संख्यात्मक प्रतिनिधित्व) उत्पादन गर्छ।

- **Diffused attention** एउटा मोडेल हो जसले embedding बाट छवि उत्पादन गर्छ। DALL-E लाई छवि र पाठको dataset मा तालिम दिइएको छ र यसले पाठ विवरणबाट छवि उत्पादन गर्न सक्छ। उदाहरणका लागि, DALL-E ले टोपी लगाएको बिरालो, वा mohawk भएको कुकुरको छवि उत्पादन गर्न सक्छ।

### Midjourney

Midjourney पनि DALL-E जस्तै काम गर्छ, यसले पाठ prompt बाट छवि उत्पादन गर्छ। Midjourney ले पनि “टोपी लगाएको बिरालो” वा “mohawk भएको कुकुर” जस्ता prompt प्रयोग गरेर छवि उत्पादन गर्न सक्छ।

## DALL-E र Midjourney कसरी काम गर्छन्

पहिले, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst)। DALL-E एक Generative AI मोडेल हो जुन transformer architecture मा आधारित छ, जसमा *autoregressive transformer* प्रयोग गरिएको छ।

*Autoregressive transformer* ले मोडेलले कसरी पाठ विवरणबाट छवि उत्पादन गर्छ भन्ने निर्धारण गर्छ, यसले एक पटकमा एक पिक्सेल उत्पादन गर्छ, अनि उत्पन्न भएका पिक्सेलहरू प्रयोग गरेर अर्को पिक्सेल उत्पादन गर्छ। यो प्रक्रिया neural network का धेरै तहहरू हुँदै जान्छ, जबसम्म छवि पूरा हुँदैन।

यस प्रक्रियाबाट, DALL-E ले उत्पन्न हुने छविमा गुण, वस्तु, विशेषता आदि नियन्त्रण गर्न सक्छ। यद्यपि, DALL-E 2 र 3 ले उत्पन्न हुने छविमा अझ बढी नियन्त्रण दिन्छ,


## आफ्नो पहिलो इमेज जेनेरेशन एप्लिकेसन बनाउँदै

एउटा इमेज जेनेरेशन एप्लिकेसन बनाउन के चाहिन्छ त? तपाईंलाई यी लाइब्रेरीहरू चाहिन्छ:

- **python-dotenv**, तपाईंलाई यो लाइब्रेरी प्रयोग गर्न सिफारिस गरिन्छ ताकि तपाईंका गोप्य जानकारीहरू *.env* फाइलमा राख्न सक्नुहोस्, कोडबाट टाढा।
- **openai**, यो लाइब्रेरी तपाईंले OpenAI API सँग अन्तरक्रिया गर्न प्रयोग गर्नुहुन्छ।
- **pillow**, Python मा इमेजहरूसँग काम गर्न।
- **requests**, HTTP अनुरोधहरू गर्न सहयोग गर्न।

1. *.env* नामको फाइल निम्न सामग्रीसहित बनाउनुहोस्:

    ```text

    pillow
    requests
    ```


1. अब, भर्चुअल वातावरण सिर्जना गर्नुहोस् र लाइब्रेरीहरू इन्स्टल गर्नुहोस्:

1. *app.py* नामको फाइलमा तलको कोड थप्नुहोस्:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # import dotenv
    dotenv.load_dotenv()
    
    # Get endpoint and key from environment variables
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # Assign the API version (DALL-E is currently supported for the 2023-06-01-preview API version only)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # Create an image by using the image generation API
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # Set the directory for the stored image
        image_dir = os.path.join(os.curdir, 'images')
    
        # If the directory doesn't exist, create it
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # Initialize the image path (note the filetype should be png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # Retrieve the generated image
        image_url = generation_response["data"][0]["url"]  # extract image URL from response
        generated_image = requests.get(image_url).content  # download the image
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # Display the image in the default image viewer
        image = Image.open(image_path)
        image.show()
    
    # catch exceptions
    except openai.InvalidRequestError as err:
        print(err)

    ```

### image generate गर्ने बारे थप जानकारी

अब image generate गर्ने कोडलाई अझ विस्तारमा हेरौं:

```python
generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
        size='1024x1024',
        n=2,
        temperature=0,
    )
```

- **prompt** भनेको त्यो text हो जुन image generate गर्न प्रयोग हुन्छ। यहाँ हामी "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" भन्ने prompt प्रयोग गर्दैछौं।
- **size** भनेको बनाइने image को साइज हो। यहाँ हामी 1024x1024 pixels को image बनाउँदैछौं।
- **n** भनेको कति वटा images generate गर्ने भन्ने हो। यहाँ हामी दुई वटा images बनाउँदैछौं।
- **temperature** भनेको Generative AI model को output कति random हुने भन्ने नियन्त्रण गर्ने parameter हो। यो 0 देखि 1 सम्मको मान हो, जहाँ 0 ले deterministic output दिन्छ र 1 ले random output दिन्छ। default value 0.7 हो।

images सँग अझ धेरै कुरा गर्न सकिन्छ, जुन हामी अर्को section मा छलफल गर्नेछौं।

## image generation का थप क्षमताहरू

तपाईंले तलका कामहरू पनि गर्न सक्नुहुन्छ:

- **Edit गर्नुहोस्**। कुनै existing image, mask र prompt दिएर image परिवर्तन गर्न सकिन्छ। उदाहरणका लागि, image को कुनै भागमा केही थप्न सकिन्छ। हाम्रो bunny image लाई सोचौं, तपाईं bunny लाई टोपी लगाउन सक्नुहुन्छ। त्यसका लागि image, mask (परिवर्तन गर्नुपर्ने भाग चिनाउने) र के गर्नुपर्छ भन्ने text prompt दिनुपर्छ।

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    base image मा केवल खरायो हुनेछ, तर अन्तिम image मा खरायोको टाउकोमा टोपी हुनेछ।
