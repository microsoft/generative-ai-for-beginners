## DALL-E र Midjourney के हुन्? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) र [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) दुई सबैभन्दा लोकप्रिय छवि उत्पादन मोडेलहरू हुन्, तिनीहरूले प्रोम्प्टहरू प्रयोग गरेर छविहरू उत्पादन गर्न अनुमति दिन्छन्।

### DALL-E

हामी DALL-E बाट सुरु गरौं, जुन Generative AI मोडेल हो जसले पाठ वर्णनबाट छविहरू उत्पादन गर्दछ। 

> [DALL-E दुई मोडेलहरूको संयोजन हो, CLIP र प्रसारित ध्यान](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)।  

- **CLIP**, एक मोडेल हो जसले छवि र पाठबाट डेटाका संख्यात्मक प्रतिनिधित्वहरू (एनबेडिङ्स) उत्पादन गर्दछ।  

- **प्रसारित ध्यान**, एक मोडेल हो जसले एनबेडिङ्सबाट छविहरू उत्पादन गर्दछ। DALL-E लाई छवि र पाठको डाटासेटमा तालिम दिइएको छ र यसले पाठ वर्णनबाट छविहरू उत्पादन गर्न सक्छ। उदाहरणका लागि, DALL-E ले टोपी लगाएको बिरालो वा मोहक भएको कुकुरका छविहरू बनाउन सक्छ। 

### Midjourney
 
Midjourney पनि DALL-E जस्तै काम गर्दछ, यो पाठ प्रोम्प्टहरूबाट छवि उत्पादन गर्छ। Midjourney पनि "टोपी लगाएको बिरालो" वा "मोहक भएको कुकुर" जस्ता प्रोम्प्ट प्रयोग गरेर छविहरू बनाउन सक्छ।

## DALL-E र Midjourney कसरी काम गर्छन् 

पहिले, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst)। DALL-E एउटा Generative AI मोडेल हो जुन ट्रान्सफर्मर आर्किटेक्चरमा आधारित छ र *autoregressive transformer* हो।  

एक *autoregressive transformer* ले मोडेलले कसरी पाठ वर्णनबाट छविहरू उत्पादन गर्छ भनेर परिभाषित गर्छ, यो एक पटकमा एउटा पिक्सेल उत्पादन गर्छ, र त्यसपछि उत्पादित पिक्सेलहरूलाई प्रयोग गरी अर्को पिक्सेल उत्पादन गर्छ। यो कार्यलाई न्यूरल नेटवर्कका धेरै तहहरुमा पार गर्दै गर्दा सम्पूर्ण छवि तयार हुन्छ।  

यस प्रक्रियाले DALL-E लाई छवि उत्पादन गर्दा विशेषताहरू, वस्तुहरू, प्रकार्यहरू र थपलाई नियन्त्रण गर्न अनुमति दिन्छ। यद्यपि, DALL-E 2 र 3 मा उत्पादन गरिएको छवि माथि अझ बढी नियन्त्रण छ, 


## तपाईंको पहिलो छवि उत्पादन अनुप्रयोग बनाउँदै

त हामीलाई छवि उत्पादन अनुप्रयोग बनाउन के के चाहिन्छ? तपाईंलाई निम्न पुस्तकालयहरू चाहिन्छ:

- **python-dotenv**, यो पुस्तकालय तपाईंका गोप्य कुराहरू कोडबाट टाढा *.env* फाइलमा राख्न अत्यन्त सिफारिस गरिएको छ।
- **openai**, यो पुस्तकालय OpenAI API सँग अन्तरक्रिया गर्न प्रयोग हुनेछ।
- **pillow**, पाइथनमा छविहरूसँग काम गर्नका लागि।
- **requests**, HTTP अनुरोधहरू बनाउन मद्दत गर्नको लागि।

यदि अहिलेसम्म गरिएको छैन भने, Azure OpenAI स्रोत र मोडेल बनाउनको लागि [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) पृष्ठमा निर्देशनहरू पालना गर्नुहोस्। मोडेल रूपमा **gpt-image-1** रोज्नुहोस् (हालको पुस्ता Azure OpenAI छवि मोडेल; DALL-E 3 पुरानो हो र नयाँ परिनियोजनहरूको लागि उपलब्ध छैन)।

1. *.env* नामको फाइल निम्न सामग्रीसहित बनाउनुहोस्:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    यो जानकारी तपाईंको स्रोतको "Deployments" खण्डमा Microsoft Foundry पोर्टलमा फेला पार्न सकिन्छ।



1. माथिका लाइब्रेरीहरूलाई *requirements.txt* नामको फाइलमा यसरी सङ्कलन गर्नुहोस्:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. त्यसपछि, भर्चुअल वातावरण सिर्जना गरी लाइब्रेरीहरू स्थापना गर्नुहोस्:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> विंडोजका लागि, तपाईँको भर्चुअल वातावरण सिर्जना र सक्रिय गर्न तलको आदेशहरू प्रयोग गर्नुहोस्:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* नामको फाइलमा तलको कोड थप्नुहोस्:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv आयात गर्नुहोस्
    dotenv.load_dotenv()

    # Azure OpenAI सेवा क्लाइन्ट कन्फिगर गर्नुहोस्
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # छवि जेनेरेशन API प्रयोग गरेर छवि सिर्जना गर्नुहोस्
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # यहाँ आफ्नो प्रॉम्प्ट पाठ प्रविष्ट गर्नुहोस्
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # संग्रहित छविका लागि निर्देशनिका सेट गर्नुहोस्
        image_dir = os.path.join(os.curdir, 'images')
    
        # यदि निर्देशनिका अवस्थित छैन भने, यसलाई सिर्जना गर्नुहोस्
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # छवि पथ प्रारम्भ गर्नुहोस् (फाइल प्रकार png हुनुपर्छ भन्ने कुरा नोट गर्नुहोस्)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # सिर्जना गरिएको छवि प्राप्त गर्नुहोस्
        image_url = generation_response.data[0].url  # प्रतिक्रिया बाट छवि URL निकाल्नुहोस्
        generated_image = requests.get(image_url).content  # छवि डाउनलोड गर्नुहोस्
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # पूर्वनिर्धारित छवि दर्शकमा छवि प्रदर्शन गर्नुहोस्
        image = Image.open(image_path)
        image.show()
    
    # अपवादहरू समात्नुहोस्
    except openai.BadRequestError as err:
        print(err)

    ```

यस कोडलाई व्याख्या गरौं:

- पहिले, हामीलाई आवश्यक पुस्तकालयहरू आयात गर्छौं, जसमा OpenAI पुस्तकालय, dotenv पुस्तकालय, requests पुस्तकालय, र Pillow पुस्तकालय समावेश छन्।

    ```

- त्यसपछि, हामी Azure OpenAI सेवा ग्राहक सेटअप गर्छौं।

    ```

- त्यसपछि, हामी छवि उत्पन्न गर्छौं:

        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # यहाँ तपाईंको प्रॉम्प्ट पाठ प्रविष्ट गर्नुहोस्

- अन्त्यमा, हामी छवि खोल्छौं र यसलाई देखाउन सामान्य छवि दर्शक प्रयोग गर्छौं:

### छवि उत्पन्न गर्ने बारे थप विवरण

अब हैन, त्यस कोडलाई हेर्नुहोस् जसले छवि उत्पन्न गर्छ:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # आफ्नो प्रॉम्प्ट टेक्स्ट यहाँ प्रविष्ट गर्नुहोस्
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt** भनेको छवि उत्पन्न गर्न प्रयोग गरिएको पाठ प्रेरेणा हो। यस मामले, हामी "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" प्रेरेणा प्रयोग गर्दैछौं।
- **size** भनेको उत्पन्न गरिएको छविको आकार हो। यस अवस्थामा, हामी 1024x1024 पिक्सेलको छवि उत्पन्न गर्दैछौं।
- **n** भनेको उत्पन्न गरिएका छविहरुको संख्या हो। यस अवस्थामा, हामी एक छवि उत्पन्न गर्दैछौं।

छविसँग गर्ने थप कुराहरू छन् जुन हामी अर्को खण्डमा समेट्नेछौं।

## छवि उत्पन्न गर्ने थप क्षमता

अहिलेसम्म तपाइँले कसरि थोरै Python कोड प्रयोग गरेर छवि उत्पन्न गर्न सकिन्छ हेर्नुभयो। यद्यपि, छविसँग गर्ने अरु धेरै कुराहरू छन्।

तपाईले तलका कुराहरू पनि गर्न सक्नुहुन्छ:

- **सम्पादनहरू गर्नुहोस्**। पहिलेको छविलाई मास्क र प्रेरेणाको साथ प्रदान गरेर तपाईं छवि परिवर्तन गर्न सक्नुहुन्छ। उदाहरणको लागि, तपाईं कुनै छविको भागमा केहि थप्न सक्नुहुन्छ। हाम्रो बन्नी छविको कल्पना गर्नुहोस्, तपाईं बन्नीमा टोपी थप्न सक्नुहुन्छ। त्यसका लागि तपाईँले छवि, मास्क (परिवर्तन हुनुपर्ने क्षेत्र चिन्नका लागि) र के गर्नुपर्छ भन्ने पाठ प्रेरेणा दिनुहुनेछ।

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    आधार छवि केवल खरायो मात्र समावेश गर्दछ तर अन्तिम छविमा खरायोमा टोपी हुनेछ।
