# छवि सिर्जना अनुप्रयोगहरू निर्माण गर्दै (एजुर ओपनएआई)

LLMहरूले केवल पाठ सिर्जनामा सिमित छैनन्। तपाईं पाठ वर्णनबाट छविहरू पनि सिर्जना गर्न सक्नुहुन्छ। मिडटेक, वास्तुकला, पर्यटन, खेल विकास, मार्केटिंग, र थप क्षेत्रमा छविहरू एक मोडालिटीको रूपमा उपयोगी हुन्छन्। यस पाठमा हामी आजका **GPT Image** मोडेलहरूसँग माइक्रोसफ्ट फाउन्ड्रीमा काम गर्छौं।

## सिकाइ लक्ष्यहरू

यस पाठको अन्त्यमा तपाईं सक्षम हुनुहुनेछ:

- छवि सिर्जना के हो र यो कहाँ उपयोगी छ व्याख्या गर्न।
- `gpt-image` मोडेल परिवार बुझ्न र यसले पुराना DALL·E मोडेलहरू भन्दा कसरी फरक छ बुझ्न।
- छवि सिर्जना अनुप्रयोग बनाउने र छविहरू सम्पादन गर्ने।

## छवि सिर्जना के हो?

छवि सिर्जना मोडेलहरूले पाठ प्रॉम्प्टबाट छविहरू सिर्जना गर्छन्। आधुनिक मोडेलहरू जस्तै `gpt-image` प्रशिक्षणको समयमा पाठ र छविहरूबीचको सम्बन्ध सिक्छन्, अनि क्रमिक रूपमा यादृच्छिक आवाजलाई तपाईंको प्रॉम्प्टसँग मेल खाने छविमा परिणत गर्दछ।

दुई परिचित छवि मोडेल परिवारहरू छन्:

- **`gpt-image` (OpenAI)** — यो पाठमा प्रयोग गरिएको वर्तमान पुस्ता। यसले पाठ-देखि-छवि सिर्जना र छवि सम्पादन (मास्क सहित इनपेन्टिंग) समर्थन गर्दछ।
- **Midjourney** — आफ्नै सेवा र Discord-आधारित कार्यप्रवाह भएको लोकप्रिय तेस्रो-पक्ष मोडेल।

> पुराना OpenAI छवि मोडेलहरू — **DALL·E 2** र **DALL·E 3** — पुराना छन्। DALL·E 3 नयाँ परिनियोजनहरूका लागि उपलब्ध छैन, र `create_variation` जस्ता सुविधाहरू केवल DALL·E 2 मा थिए। नयाँ अनुप्रयोगहरूको लागि `gpt-image` मोडेलहरू प्रयोग गर्नुहोस्।

माइक्रोसफ्ट फाउन्ड्रीमा, **`gpt-image-2`** नवीनतम र सबैभन्दा सक्षम छवि मोडेल हो र यो सिफारिस गरिएको पूर्वनिर्धारित हो। `gpt-image-1.5` र `gpt-image-1-mini` पनि सामान्य उपलब्ध छन्।

> **महत्वपूर्ण:** `gpt-image` मोडेलहरूले सिर्जित छविलाई **base64** (`b64_json`) रूपमा दिन्छन्, URL रूपमा होइन। तपाईंको कोडले base64 स्ट्रिङलाई बाइट्समा डिकोड गरी सुरक्षित गर्छ — डाउनलोड गर्न कुनै छवि URL हुँदैन।


## तपाईंको पहिलो इमेज जेनेरेशन अनुप्रयोग निर्माण गर्दै

तपुर कस्तो हुन्छ इमेज जेनेरेशन अनुप्रयोग बनाउन? तपाईँलाई यी लाइब्रेरीहरूको आवश्यकता छ:

- **python-dotenv**, तपाईंलाई यो लाइब्रेरी प्रयोग गर्न अत्यन्त सिफारिस गरिन्छ ताकि तपाईंका गोप्य जानकारीहरू कोडबाट टाढा *.env* फाइलमा राख्न सकियोस्।
- **openai**, यो लाइब्रेरी हो जुन तपाईँले OpenAI API सँग अन्तर्क्रिया गर्न प्रयोग गर्नुहुनेछ।
- **pillow**, Python मा इमेजहरूसँग काम गर्न।

यदि पहिल्यै नभएको हो भने, [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) पृष्ठमा दिइएका निर्देशनहरू अनुसरण गरेर Microsoft Foundry स्रोत र मोडेल बनाउनुहोस्। मोडेलको रूपमा **gpt-image-2** चयन गर्नुहोस् (सबैभन्दा नयाँ Azure OpenAI इमेज मोडेल; DALL·E पुरानो छ)।

1. निम्न सामग्रीसहित *.env* फाइल बनाउनुहोस्:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    उक्त जानकारी Microsoft Foundry पोर्टलमा "Deployments" सेक्सनमा तपाईँको स्रोतको लागि फेला पार्न सकिन्छ।



1. माथिका पुस्तकालयहरूलाई *requirements.txt* नामक फाइलमा यसरी सङ्कलन गर्नुहोस्:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. त्यसपछि, भर्चुअल वातावरण सिर्जना गरी पुस्तकालयहरू स्थापना गर्नुहोस्:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> विन्डोजको लागि, तपाईले आफ्नो भर्चुअल वातावरण सिर्जना र सक्रिय गर्न निम्न आदेशहरू प्रयोग गर्नुहोस्:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* नामको फाइलमा तल दिएको कोड थप्नुहोस्:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv आयात गर्नुहोस्
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) क्लाइन्ट कन्फिगर गर्नुहोस्।
    # इमेज मोडेलहरूले नयाँ API संस्करण आवश्यक पर्छ — तपाईंको मोडेलले चाहिने संस्करणका लागि Foundry कागजात जाँच गर्नुहोस्।
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # इमेज जेनेरेसन API प्रयोग गरेर तस्बिर सिर्जना गर्नुहोस्
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # तपाइँको प्रॉम्प्ट टेक्स्ट यहाँ राख्नुहोस्
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # जस्तै "gpt-image-2"
        )
        # स्टोर गरिएको तस्बिरको लागि फोल्डर सेट गर्नुहोस्
        image_dir = os.path.join(os.curdir, 'images')

        # यदि फोल्डर छैन भने, यसलाई सिर्जना गर्नुहोस्
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # इमेज पथ सुरु गर्नुहोस् (ध्यान राख्नुहोस् फाइल प्रकार png हुनु पर्छ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image मोडेलहरूले तस्बिर URL होइन, base64 (b64_json) मा फर्काउँछन्
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # तस्बिरलाई डिफल्ट इमेज भ्युअरमा देखाउनुहोस्
        image = Image.open(image_path)
        image.show()

    # अपवादहरू समात्नुहोस्
    except BadRequestError as err:
        print(err)

    ```

यस कोडको व्याख्या गरौं:

- पहिलो, हामी लाइब्रेरीहरू इम्पोर्ट गर्छौं जुन आवश्यक छ, जसमा OpenAI लाइब्रेरी, dotenv लाइब्रेरी, base64 मोड्युल, र Pillow लाइब्रेरी समावेश छ।

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- त्यसपछि, हामी *.env* फाइलबाट वातावरण चरहरू लोड गर्छौं।

    ```python
    # dotenv आयात गर्नुहोस्
    dotenv.load_dotenv()
    ```

- त्यसपछि, हामी Azure OpenAI (Microsoft Foundry) क्लाइन्ट कन्फिगर गर्छौं।

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- अगाडि, हामी तस्बिर उत्पादन गर्छौं:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # यहाँ आफ्नो प्रॉम्प्ट टेक्स्ट प्रविष्ट गर्नुहोस्
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` मोडेलले तस्बिरलाई `data[0].b64_json` मा **base64** स्ट्रिङको रूपमा फिर्ता गर्छ। हामी यसलाई बाइट्समा डिकोड गरेर फाइलमा लेख्छौं — डाउनलोड गर्न URL हुँदैन।

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- अन्तमा, हामी तस्बिर खोल्छौं र यसलाई देखाउन सामान्य तस्बिर हेर्ने कार्यक्रम प्रयोग गर्छौं:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### तस्बिर उत्पादन गर्ने बिषयमा थप विवरण

`images.generate` का प्यारामिटरहरू हेरौं:

- **prompt**, तस्बिर बनाउन प्रयोग गरिएको टेक्स्ट प्रॉम्प्ट। यहाँ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" छ।
- **size**, उत्पादित तस्बिरको आकार हो (जस्तै `1024x1024`, `1536x1024`, `1024x1536`, वा `"auto"`).
- **n**, बनाइने तस्बिरहरूको संख्या हो। यहाँ हामी एक तस्बिर बनाउँछौं।
- **model**, तपाईंको तस्बिर मोडेल डिप्लोइमेंट नाम हो (जस्तै `gpt-image-2`)।

> तस्बिर मोडेलहरूले `temperature` प्यारामिटर लिंदैन — त्यो टेक्स्ट-निर्माण नियन्त्रण हो। विविधता प्राप्त गर्न, API फेरि कल गर्नुहोस्; विविधता घटाउन, प्रॉम्प्टलाई थप स्पष्ट बनाउनुहोस्।

## तस्बिर उत्पादनका थप क्षमता

तपाईंले केही लाइनको Python कोडबाट तस्बिर कसरी बनाउने हेर्नुभयो। `gpt-image` मोडेलहरूले पहिलेको तस्बिर **सम्पादन** पनि गर्न सक्छ। एक तस्बिर, विकल्पमा **mask** (जे क्षेत्र परिवर्तन गर्नुपर्छ त्यो मार्क गर्ने), र प्रॉम्प्ट दिँदा, तपाईं तस्बिरको भाग परिवर्तन गर्नसक्नुहुन्छ — जस्तै, हाम्रो खर्बुजा खरायोमा टोपी थप्नु।

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# सम्पादनहरू पनि base64 रूपमा फर्काइन्छन्
edited_image = base64.b64decode(response.data[0].b64_json)
```

आधार तस्बिरमा केवल खरायो छ; अन्तिम तस्बिरमा खरायोमा टोपी छ।
