### 2. 匯入輔助庫並實例化憑證


```python
import os
from openai import AzureOpenAI
import numpy as np
from dotenv import load_dotenv
load_dotenv()

#validate data inside .env file

client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  # this is also the default, it can be omitted
  api_version = "2023-05-15"
  )
```

## 4. 提示設計  

「大型語言模型的魔力在於，透過在大量文本中訓練以最小化此預測錯誤，模型最終學會了對這些預測有用的概念。例如，它們學會了像」(1):

* 如何拼寫

#### 如何控制大型語言模型  
「在所有輸入大型語言模型的資料中，最具影響力的無疑是文字提示(1)。

大型語言模型可以透過幾種方式被提示產生輸出：

- 指令：告訴模型你想要什麼
- 補全：誘導模型完成你想要的開頭部分
- 示範：向模型展示你想要的內容，方式包括：
- 在提示中給出幾個範例
- 在微調訓練資料集中給出數百或數千個範例



#### 創建提示的三個基本指導原則：

**展示並說明**。透過指令、範例或兩者結合，清楚表達你想要什麼。如果你想讓模型將一串項目按字母順序排序，或根據情感對段落進行分類，就要讓它知道這是你的需求。

**提供高品質資料**。如果你想建立分類器或讓模型遵循某種模式，確保有足夠的範例。務必校對你的範例——模型通常足夠聰明，能看穿基本拼寫錯誤並給出回應，但它也可能假設這是故意的，這會影響回應。

**檢查你的設定。** 溫度（temperature）和 top_p 設定控制模型生成回應的確定性。如果你要求的回應只有一個正確答案，則應將這些設定調低。如果你想要更多元的回應，則可以將它們調高。人們對這些設定最常犯的錯誤是誤以為它們是「聰明度」或「創造力」的控制開關。


來源：https://learn.microsoft.com/azure/ai-services/openai/overview

### 5. 提交！


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.chat.completions.create(
  model=model,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```

## Embeddings
本節將展示如何檢索嵌入向量，以及如何尋找單詞、句子和文件之間的相似性。要運行以下筆記本，您需要部署一個以 `text-embedding-ada-002` 為基礎模型的模型，並在 .env 文件中使用 `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` 變量設置其部署名稱。


### 模型分類法 - 選擇嵌入模型

**模型分類法**：{family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (嵌入模型系列)  
{capability} --> ada             (所有其他嵌入模型將於2024年退役)  
{input-type} --> n/a             (僅為搜尋模型指定)  
{identifier} --> 002             (版本002)  

model = 'text-embedding-ada-002'


  > [!注意] 如果在 Codespaces 或 Devcontainer 中運行此筆記本，則以下步驟不是必需的


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
