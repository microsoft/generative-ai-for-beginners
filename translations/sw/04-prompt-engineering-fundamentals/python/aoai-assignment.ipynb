# Utangulizi wa Uhandisi wa Prompti
Uhandisi wa prompti ni mchakato wa kubuni na kuboresha prompti kwa ajili ya kazi za usindikaji wa lugha ya asili. Inajumuisha kuchagua prompti sahihi, kurekebisha vigezo vyake, na kutathmini utendaji wake. Uhandisi wa prompti ni muhimu kwa kufanikisha usahihi na ufanisi mkubwa katika mifano ya NLP. Katika sehemu hii, tutaangazia misingi ya uhandisi wa prompti tukitumia mifano ya OpenAI kwa ajili ya uchunguzi.


### Zozi 1: Utengenezaji wa Tokeni
Chunguza Utengenezaji wa Tokeni ukitumia tiktoken, kichakataji tokeni cha haraka cha chanzo huru kutoka OpenAI
Angalia [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) kwa mifano zaidi.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Zoef 2: Thibitisha Usanidi wa Funguo ya API ya OpenAI

Endesha msimbo ulio hapa chini ili kuthibitisha kuwa mlango wako wa OpenAI umetayarishwa vizuri. Msimbo huu unajaribu tu ombi rahisi na kuthibitisha ukamilisho. Ingizo `oh say can you see` inapaswa kumalizika kwa mistari kama `by the dawn's early light..`


````python
# Uses the OpenAI client against the Azure OpenAI (Microsoft Foundry) v1 endpoint
# with the Responses API. See https://aka.ms/openai/start

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### Zoefu la 3: Ujengaji
Chunguza kinachotokea unapoomba LLM irudishe ukamilishaji kwa kichocheo kuhusu mada ambayo huenda haipo, au kuhusu mada ambazo huenda haijui kwa sababu zilikuwa nje ya seti yake ya mafunzo ya awali (ya karibuni zaidi). Tazama jinsi jibu linavyobadilika ikiwa utajaribu kichocheo tofauti, au mfano tofauti.


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### Mazoezi 4: Kulingana na Maelekezo 
Tumia variable ya "text" kuweka yaliyomo kuu 
na variable ya "prompt" kutoa maelekezo yanayohusiana na yaliyomo kuu.

Hapa tunaomba modeli ifanye muhtasari wa maandishi kwa mwanafunzi wa darasa la pili


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### Zoezi 5: Ombi ngumu  
Jaribu ombi lenye ujumbe wa mfumo, mtumiaji na msaidizi  
Mfumo unaweka muktadha wa msaidizi  
Ujumbe wa Mtumiaji na Msaidizi hutoa muktadha wa mazungumzo yenye mizunguko mingi  

Angalia jinsi tabia ya msaidizi imewekwa kuwa "mchokozi" katika muktadha wa mfumo.  
Jaribu kutumia muktadha wa tabia tofauti. Au jaribu mfululizo tofauti wa ujumbe wa ingizo/zao  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
