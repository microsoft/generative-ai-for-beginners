# Tengeneza programu za kizazi cha maandishi

Umeona hadi sasa kupitia mtaala huu kuna dhana za msingi kama vile maagizo na hata taaluma nzima iitwayo "handisi ya maagizo". Zana nyingi unazoweza kuingiliana nazo kama ChatGPT, Office 365, Microsoft Power Platform na zingine, zinakuunga mkono kutumia maagizo kufanikisha jambo.

Ili kuongeza uzoefu kama huo katika programu, unahitaji kuelewa dhana kama maagizo, makamilisho na kuchagua maktaba ya kufanya nayo kazi. Hiyo ndiyo hasa utakayojifunza katika sura hii.

## Utangulizi

Katika sura hii, utakuwa:

- Kujifunza kuhusu maktaba ya openai na dhana zake za msingi.
- Kujenga programu ya kizazi cha maandishi ukitumia openai.
- Kuelewa jinsi ya kutumia dhana kama agizo, joto, na tokeni kujenga programu ya kizazi cha maandishi.

## Malengo ya kujifunza

Mwisho wa somo hili, utaweza:

- Eleza ni nini programu ya kizazi cha maandishi.
- Tengeneza programu ya kizazi cha maandishi ukitumia openai.
- Sanidi programu yako kutumia tokeni zaidi au kidogo na pia badilisha joto, kwa matokeo tofauti.

## Ni nini programu ya kizazi cha maandishi?

Kawaida unapotengeneza programu ina aina fulani ya kiolesura kama ifuatavyo:

- Inayotumia amri. Programu za consoli ni programu za kawaida ambapo unaandika amri na hufanya kazi. Kwa mfano, `git` ni programu inayotumia amri.
- Kiolesura cha mtumiaji (UI). Baadhi ya programu zina kiolesura cha mtumiaji cha picha (GUI) ambapo unaweza kubonyeza vifungo, kuingiza maandishi, kuchagua chaguzi na zaidi.

### Programu za consoli na UI zina vikwazo

Linganisha na programu inayotumia amri ambapo unaandika amri:

- **Ina vikwazo**. Huwezi kuandika amri yoyote ile, bali zile zinazoungwa mkono na programu.
- **Ina lugha maalum**. Baadhi ya programu zinaunga mkono lugha nyingi, lakini kwa kawaida programu imetengenezwa kwa lugha maalum, hata kama unaweza kuongeza lugha zaidi.

### Manufaa ya programu za kizazi cha maandishi

Basi programu ya kizazi cha maandishi ni tofauti vipi?

Katika programu ya kizazi cha maandishi, una uhuru zaidi, huna kikomo cha seti ya amri au lugha maalum ya kuingiza. Badala yake, unaweza kutumia lugha ya asili kuingiliana na programu. Faida nyingine ni kwamba kwa kuwa tayari unaingiliana na chanzo cha data kilichofunzwa kwa msururu mkubwa wa taarifa, wakati programu ya kawaida inaweza kuwa na kikomo cha kile kilicho kwenye hifadhidata.

### Naweza kutengeneza nini kwa programu ya kizazi cha maandishi?

Kuna mambo mengi unayoweza kutengeneza. Kwa mfano:

- **Roboti wa mazungumzo**. Roboti wa mazungumzo anayejibu maswali kuhusu mada, kama kampuni yako na bidhaa zake anaweza kuwa chaguo zuri.
- **Msaidizi**. LLMs ni nzuri katika mambo kama kunukuu maandishi, kupata maarifa kutoka maandishi, kutengeneza maandishi kama wasifu na zaidi.
- **Msaidizi wa msimbo**. Kulingana na mfano wa lugha unayotumia, unaweza kutengeneza msaidizi wa msimbo anayekusaidia kuandika msimbo. Kwa mfano, unaweza kutumia bidhaa kama GitHub Copilot pamoja na ChatGPT kukusaidia kuandika msimbo.

## Ninaweza kuanza vipi?

Vizuri, unahitaji kupata njia ya kuunganishwa na LLM ambayo kawaida inahusisha njia mbili zifuatazo:

- Tumia API. Hapa unaunda maombi ya wavuti kwa kutumia agizo lako na kurudisha maandishi yaliyozalishwa.
- Tumia maktaba. Maktaba husaidia kufunga miito ya API na kuifanya iwe rahisi kutumia.

## Zalisha maandishi

Njia ya kuzaa maandishi ni kutumia njia `responses.create`. Hapa kuna mfano:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Katika msimbo uliotajwa, tunaunda kitu cha majibu na kuingiza mfano tunayotaka kutumia pamoja na agizo. Kisha tunaandika maandishi yaliyozalishwa.

### Majibu ya mazungumzo

Hadi sasa, umeona jinsi tunavyoweza kuzaa maandishi kutoka kwa agizo moja. API ya Majibu pia ni nzuri kwa roboti za mazungumzo - unatoa orodha ya ujumbe kama `input` kujenga mazungumzo. Hapa kuna mfano wa kutumia hii:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Zaidi kuhusu utendakazi huu katika sura inayoifuata.

Sasa tumepata jinsi ya kusanidi na kuanzisha huduma ya OpenAI, ni wakati wa kutengeneza programu yako ya kwanza ya kizazi cha maandishi. Ili kutengeneza programu yako, fuata hatua hizi:

1. Tafuta kigezo `prompt` na badilisha msimbo wake kuwa ifuatayo:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Neno kuhusu urefu wa tokeni**. Tunapaswa kuzingatia tokeni ngapi tunahitaji kuunda maandishi tunayoyataka. Tokeni zinagharimu pesa, hivyo inapowezekana, tunapaswa kuwa na uangalifu kuhusu idadi ya tokeni tunazotumia. Kwa mfano, tunaweza kubadilisha ombi ili kutumia tokeni chache?

   Kubadilisha tokeni zinazotumika, unaweza kutumia vigezo `max_output_tokens`. Kwa mfano, ikiwa unataka kutumia tokeni 100, utatumia:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Jaribu na joto la joto**. Joto ni kitu ambacho hatujazungumzia hadi sasa lakini ni muktadha muhimu kwa jinsi programu yetu inavyofanya kazi. Joto kubwa hutoa matokeo yasiyotabirika zaidi.    Kubadilisha joto, unaweza kutumia vigezo `temperature`. Kwa mfano, ikiwa unataka kutumia joto 0.5, utatumia:

