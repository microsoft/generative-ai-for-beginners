# Jenga programu za kizazi cha maandishi

Umeona hadi sasa kupitia mtaala huu kwamba kuna dhana za msingi kama vile prompts na hata taaluma nzima inayoitwa "prompt engineering". Zana nyingi unaweza kuingiliana nazo kama ChatGPT, Office 365, Microsoft Power Platform na nyingine zaidi, zinakusaidia kutumia prompts kutekeleza jambo fulani.

Ili uweze kuongeza uzoefu kama huo kwenye programu, unahitaji kuelewa dhana kama prompts, completions na kuchagua maktaba ya kufanya nayo kazi. Hapo ndipo hasa utakapojifunza katika sura hii.

## Utangulizi

Katika sura hii, utafanya:

- Jifunze kuhusu maktaba ya openai na dhana zake za msingi.
- Jenga programu ya kizazi cha maandishi ukiwa unatumia openai.
- Elewa jinsi ya kutumia dhana kama prompt, halijoto (temperature), na tokens kujenga programu ya kizazi cha maandishi.

## Malengo ya kujifunza

Mwisho wa somo hili, utaweza:

- Elezea ni nini programu ya kizazi cha maandishi.
- Jenga programu ya kizazi cha maandishi kwa kutumia openai.
- Sanidi programu yako kutumia tokens nyingi au chache na pia badilisha halijoto, kwa matokeo mbalimbali.

## Je, ni nini programu ya kizazi cha maandishi?

Kawaida unapo jenga programu huwa na aina fulani ya kiolesura kama ifuatavyo:

- Kwa amri. Programu za console ni programu kawaida ambapo unaandika amri na inatekeleza kazi. Kwa mfano, `git` ni programu ya aina ya amri.
- Kiolesura cha mtumiaji (UI). Baadhi ya programu zina kiolesura cha mtumiaji kinachoonekana (GUI) ambako una bonyeza vitufe, weka maandishi, chagua chaguzi na zaidi.

### Programu za Console na UI zina mipaka

Linganisha na programu ya amri ambapo unaandika amri: 

- **Ina mipaka**. Huwezi kuandika amri yoyote, ni zile tu ambazo programu inaweza.
- **Inategemea lugha fulani**. Baadhi ya programu zinaunga mkono lugha nyingi, lakini kwa kawaida programu imetengenezwa kwa lugha moja, hata kama unaweza kuongeza msaada wa lugha zaidi.

### Faida za programu za kizazi cha maandishi

Basi, je, programu ya kizazi cha maandishi ni tofauti vipi?

Katika programu ya kizazi cha maandishi, una uhuru zaidi, huna mipaka ya orodha ya amri au lugha maalum ya kuingiza. Badala yake, unaweza kutumia lugha ya asili kuingiliana na programu. Faida nyingine ni kwamba kwa kuwa tayari unafanya kazi na chanzo cha data kilicho fundishwa kwenye jumla kubwa ya taarifa, ikilinganishwa na programu za jadi ambazo zinaweza kuwa na mipaka kuhusu kile kilicho kwenye hifadhidata.

### Ninaweza kujenga nini kwa kutumia programu ya kizazi cha maandishi?

Kuna mambo mengi unaweza kuya jenga. Kwa mfano:

- **Chatbot**. Chatbot inayojibu maswali kuhusu mada mbalimbali, kama kampuni yako na bidhaa zake inaweza kuwa chaguo nzuri.
- **Msaidizi**. LLM ni nzuri kwa vitu kama kufupisha maandishi, kupata taarifa muhimu kutoka kwenye maandishi, kutengeneza maandishi kama vile wasifu na zaidi.
- **Msaidizi wa code**. Kutegemea mfano wa lugha unayotumia, unaweza kujenga msaidizi wa kuandika code. Kwa mfano, unaweza kutumia bidhaa kama GitHub Copilot pamoja na ChatGPT kusaidia kuandika code.

## Naweza kuanza vipi?

Kweli, unahitaji kupata njia ya kuunganishwa na LLM ambayo kwa kawaida inahitaji njia mbili zilizo:

- Tumia API. Hapa unaunda maombi ya wavuti na prompt yako na upokee maandishi yaliyo tengenezwa kurudi.
- Tumia maktaba. Maktaba husaidia kufunga maombi ya API na kuyafanya yawe rahisi kutumia.

## Maktaba/SDKs

Kuna maktaba chache maarufu za kufanya na LLM kama:

- **openai**, maktaba hii hufanya iwe rahisi kuungana na mfano wako na kutuma prompts.

Kisha kuna maktaba zinazofanya kazi kwa kiwango cha juu kama:

- **Langchain**. Langchain ni maarufu na inaunga mkono Python.
- **Semantic Kernel**. Semantic Kernel ni maktaba ya Microsoft inayounga mkono lugha C#, Python, na Java.

## Tengeneza maandishi

Njia ya kutengeneza maandishi ni kutumia njia ya `responses.create`. Hapa kuna mfano:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Katika code iliyo juu, tunaunda kitu cha majibu na tunapitisha mfano tunayotaka kutumia pamoja na prompt. Kisha tunachapisha maandishi yaliyotengenezwa.

### Majibu ya mazungumzo

Hadi sasa, umeona jinsi tunavyoweza kutengeneza maandishi kutoka prompt moja tu. Responses API pia ni sawa kwa chatbot - unatoa orodha ya ujumbe kama `input` kujenga mazungumzo. Hapa kuna mfano wa kuitumia:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Zaidi kuhusu uwezo huu katika sura inayokuja.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # chapisha jibu
    print("Shopping list:")
    print(response.output_text)

- **Neno kuhusu urefu wa tokeni**. Tunapaswa kuzingatia tokeni ngapi tunahitaji kuzalisha maandishi tunayoyataka. Tokeni zinagharimu pesa, kwa hivyo inapowezekana, tunapaswa kuwa waangalifu na idadi ya tokeni tunazotumia. Kwa mfano, tunaweza kuweka maelekezo kwa njia ambayo tunaweza kutumia tokeni chache?

   Ili kubadilisha tokeni zinazotumiwa, unaweza kutumia parameter ya `max_output_tokens`. Kwa mfano, kama unataka kutumia tokeni 100, ungefanya hivi:

    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)

- **Kujaribu na joto**. Joto ni kitu ambacho hatujazungumza hadi sasa lakini ni muktadha muhimu kwa utendaji wa programu yetu. Joto kubwa zaidi, matokeo yatakuwa ya bahati zaidi.    Ili kubadilisha joto, unaweza kutumia parameter ya `temperature`. Kwa mfano, kama unataka kutumia joto la 0.5, ungefanya hivi:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Kumbuka, ukikaribia 1.0, matokeo yatakuwa na tofauti zaidi.
