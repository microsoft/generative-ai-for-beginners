# Jenga programu za kutengeneza maandishi

Kama ulivyoona kwenye mtaala huu, kuna dhana kuu kama vile vidokezo (prompts) na hata taaluma nzima inayoitwa "prompt engineering". Zana nyingi unazoweza kutumia kama ChatGPT, Office 365, Microsoft Power Platform na nyingine nyingi, zinakuwezesha kutumia vidokezo ili kutimiza jambo fulani.

Ili uongeze uzoefu kama huu kwenye programu yako, unahitaji kuelewa dhana kama vidokezo, majibu (completions) na kuchagua maktaba ya kutumia. Hicho ndicho hasa utakachojifunza kwenye sura hii.

## Utangulizi

Katika sura hii, utajifunza:

- Kuhusu maktaba ya openai na dhana zake kuu.
- Kujenga programu ya kutengeneza maandishi kwa kutumia openai.
- Kuelewa jinsi ya kutumia dhana kama prompt, temperature, na tokens kujenga programu ya kutengeneza maandishi.

## Malengo ya kujifunza

Mwisho wa somo hili, utaweza:

- Kuelezea programu ya kutengeneza maandishi ni nini.
- Kujenga programu ya kutengeneza maandishi kwa kutumia openai.
- Kusanidi programu yako kutumia tokens nyingi au chache na pia kubadilisha temperature, ili kupata matokeo tofauti.

## Programu ya kutengeneza maandishi ni nini?

Kwa kawaida unapojenga programu huwa na aina fulani ya kiolesura kama ifuatavyo:

- Inayotegemea amri. Programu za console ni zile ambazo unaandika amri na inatekeleza kazi fulani. Kwa mfano, `git` ni programu inayotegemea amri.
- Kiolesura cha mtumiaji (UI). Baadhi ya programu zina kiolesura cha picha (GUI) ambapo unabofya vitufe, unaingiza maandishi, unachagua chaguzi na mengineyo.

### Programu za Console na UI zina mipaka

Linganisheni na programu inayotegemea amri ambapo unaandika amri:

- **Ina mipaka**. Huwezi kuandika amri yoyote, ni zile tu ambazo programu inaziruhusu.
- **Lugha maalum**. Baadhi ya programu zinaunga mkono lugha nyingi, lakini kwa kawaida programu inajengwa kwa lugha maalum, hata kama unaweza kuongeza lugha nyingine.

### Faida za programu za kutengeneza maandishi

Basi, programu ya kutengeneza maandishi inatofautianaje?

Katika programu ya kutengeneza maandishi, una uhuru zaidi, hufungwi na seti ya amri au lugha maalum ya kuingiza. Badala yake, unaweza kutumia lugha ya kawaida kuwasiliana na programu. Faida nyingine ni kwamba tayari unawasiliana na chanzo cha data kilichofunzwa na taarifa nyingi sana, wakati programu ya kawaida inaweza kuwa na mipaka ya kile kilicho kwenye hifadhidata.

### Naweza kujenga nini kwa kutumia programu ya kutengeneza maandishi?

Kuna mambo mengi unaweza kujenga. Kwa mfano:

- **Chatbot**. Chatbot inayojibu maswali kuhusu mada mbalimbali, kama kampuni yako na bidhaa zake inaweza kuwa chaguo zuri.
- **Msaidizi**. LLMs ni bora katika mambo kama kufupisha maandishi, kupata maarifa kutoka kwenye maandishi, kutengeneza maandishi kama wasifu na mengine mengi.
- **Msaidizi wa kuandika msimbo**. Kulingana na mfano wa lugha unaotumia, unaweza kujenga msaidizi wa msimbo anayekusaidia kuandika msimbo. Kwa mfano, unaweza kutumia bidhaa kama GitHub Copilot pamoja na ChatGPT kukusaidia kuandika msimbo.

## Tengeneza maandishi

Njia ya kutengeneza maandishi ni kutumia darasa la `chat.completion`. Hapa kuna mfano:


1. Tengeneza faili la *app.py* na uweke msimbo ufuatao:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = completion.choices[0].text
    prompt = "Tengeneza orodha ya manunuzi kwa mapishi yaliyotengenezwa na tafadhali usijumuishe viambato ambavyo tayari ninavyo."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
    
    # chapisha majibu
    print("Orodha ya manunuzi:")
    print(completion.choices[0].message.content)
    ```

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python

        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_tokens` is 1200. 

        ```python
        completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
        ```  
