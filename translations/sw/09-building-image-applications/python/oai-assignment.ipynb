# Kujenga Programu za Uundaji Picha (OpenAI)

Kuna zaidi ya LLMs kuliko uundaji wa maandishi tu. Unaweza pia kuunda picha kutoka kwa maelezo ya maandishi. Picha kama njia ni muhimu katika MedTech, usanifu, utalii, maendeleo ya michezo, masoko, na zaidi. Katika somo hili tunafanya kazi na mifano ya **GPT Image** ya leo kwenye jukwaa la OpenAI.

## Malengo ya kujifunza

Mwisho wa somo hili utaweza:

- Eleza ni nini uundaji wa picha ni na ni wapi unatumika.
- Elewa familia ya modeli za `gpt-image` na jinsi zinavyotofautiana na modeli za zamani za DALL·E.
- Jenga programu ya uundaji picha na hariri picha.

## Uundaji wa picha ni nini?

Mifano ya uundaji picha huunda picha kutoka kwa maelekezo ya maandishi. Modeli za kisasa kama `gpt-image` hujifunza uhusiano kati ya maandishi na picha wakati wa mafunzo, kisha hupitia hatua kwa hatua hubadilisha kelele za nasibu kuwa picha inayolingana na maelekezo yako.

Familia mbili maarufu za modeli za picha ni:

- **`gpt-image` (OpenAI)** — kizazi cha sasa kinachotumika katika somo hili. Kinaunga mkono uundaji wa picha kutoka kwa maandishi na uhariri wa picha (kuchora tena kwa kutumia kasha).
- **Midjourney** — modeli maarufu ya tatu yenye huduma yake na mtiririko kazi ulio kwenye Discord.

> Modeli za picha za zamani za OpenAI — **DALL·E 2** na **DALL·E 3** — ni za jadi. DALL·E 3 haipatikani tena kwa uanzishaji mpya, na vipengele kama `create_variation` vilikuwepo tu kwenye DALL·E 2. Tumia modeli za `gpt-image` kwa programu mpya.

> **Muhimu:** modeli za `gpt-image` hurejea picha iliyoundwa kama **base64** (`b64_json`), si kama URL. Msimbo wako huzifungua mnyororo wa base64 hadi bytes na kuihifadhi — hakuna URL ya picha ya kupakua.


## Kujenga programu yako ya kwanza ya uzalishaji picha

Basi unachohitaji ili kujenga programu ya uzalishaji picha ni nini? Unahitaji maktaba zifuatazo:

- **python-dotenv**, unashauriwa sana kutumia maktaba hii kuweka siri zako katika faili la *.env* mbali na msimbo.
- **openai**, hii ni maktaba utakayotumia kuingiliana na OpenAI API.
- **pillow**, kwa kufanya kazi na picha katika Python.


1. Unda faili la *.env* lenye maudhui yafuatayo:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. Kusanya maktaba zilizo juu katika faili iitwayo *requirements.txt* kama ifuatavyo:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. Ifuatayo, tengeneza mazingira ya kidijitali na usakinishe maktaba:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Kwa Windows, tumia amri zifuatazo kuunda na kuanzisha mazingira yako pepe:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. Ongeza msimbo ufuatao katika faili inayoitwa *app.py*:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # import dotenv
    dotenv.load_dotenv()

    # Unda kitu cha OpenAI (husoma OPENAI_API_KEY kutoka kwenye .env yako)
    client = openai.OpenAI()


    try:
        # Unda picha kwa kutumia API ya uzalishaji wa picha
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Weka maandishi ya ombi lako hapa
            size='1024x1024',
            n=1
        )
        # Weka saraka kwa ajili ya picha iliyohifadhiwa
        image_dir = os.path.join(os.curdir, 'images')

        # Ikiwa saraka haipo, iundie
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Anzisha njia ya picha (kumbuka aina ya faili inapaswa kuwa png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # mifano ya gpt-image hurudisha picha kama base64 (b64_json), sio URL
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Onyesha picha kwenye mtazamaji wa picha wa chaguo-msingi
        image = Image.open(image_path)
        image.show()

    # chukua makosa
    except openai.BadRequestError as err:
        print(err)

    ```

Hebu tuelezee msimbo huu:

- Kwanza, tunaagiza maktaba tunazohitaji, pamoja na maktaba ya OpenAI, maktaba ya dotenv, moduli ya base64, na maktaba ya Pillow.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- Baada ya hapo, tunaunda mteja, ambaye husoma ufunguo wa API kutoka kwenye fayala yako ya ``.env``.

    ```python
    # Unda kitu cha OpenAI
    client = openai.OpenAI()
    ```

- Ifuatayo, tunatengeneza picha:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Ingiza maandishi ya ombi lako hapa
        size='1024x1024',
        n=1
    )
    ```

    Mfano za `gpt-image` hurejesha picha kama mfinyo wa **base64** katika `data[0].b64_json`. Tunaufasiri kuwa biti na kuandika kwenye faili — hakuna URL ya kupakua.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- Mwishowe, tunafungua picha na kutumia mtazamaji picha wa kawaida kuionyesha:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### Maelezo zaidi juu ya kutengeneza picha

Hebu tazame vigezo vya `images.generate`:

- **model**, ni mfano wa picha wa kutumia (kwa mfano `gpt-image-1`).
- **prompt**, ni maandishi ya agizo yanayotumika kutengeneza picha. Hapa ni "Bunny kwenye farasi, akikamata pipi, kwenye uwanja ulio na ukungu ambapo hukua daffodils".
- **size**, ni ukubwa wa picha iliyotengenezwa (kwa mfano `1024x1024`, `1536x1024`, `1024x1536`, au `"auto"`).
- **n**, ni idadi ya picha zilizotengenezwa. Hapa tunatengeneza moja.

> Mifano ya picha haiwezi kutumia kipengele cha `temperature` — hicho ni udhibiti wa uundaji wa maandishi. Ili kupata utofauti, piga simu API tena; kupunguza utofauti, fanya agizo lako liwe maalum zaidi.

## Uwezo wa ziada wa utengenezaji picha

Umeona jinsi ya kutengeneza picha kwa mistari michache ya Python. Mifano ya `gpt-image` pia inaweza **kuhariri** picha iliyopo. Kwa kutoa picha, **maski** hiari (inayoonyesha eneo la kubadilisha), na agizo, unaweza kubadilisha sehemu ya picha — kwa mfano, ongeza kofia kwa sungura wetu.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# mabadiliko pia hurudishwa kama base64
edited_image = base64.b64decode(response.data[0].b64_json)
```

Picha ya msingi ina sungura tu; picha ya mwisho ina kofia kwenye sungura.
