## DALL-E na Midjourney ni nini? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) na [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) ni mifano miwili maarufu ya kuzalisha picha, huruhusu kutumia maelekezo kuzalisha picha.

### DALL-E

Tuanze na DALL-E, ambayo ni mfano wa AI inayozalisha picha kutoka kwa maelezo ya maandishi. 

> [DALL-E ni mchanganyiko wa mifano miwili, CLIP na diffused attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **CLIP**, ni mfano unaozalisha embeddings, ambayo ni uwakilishi wa nambari wa data, kutoka kwa picha na maandishi.  

- **Diffused attention**, ni mfano unaozalisha picha kutoka kwa embeddings. DALL-E imefunzwa kwa seti ya data ya picha na maandishi na inaweza kutumika kuzalisha picha kutoka kwa maelezo ya maandishi. Kwa mfano, DALL-E inaweza kutumika kuzalisha picha ya paka aliyevaa kofia, au mbwa mwenye mohawk. 

### Midjourney
 
Midjourney hufanya kazi kwa njia inayofanana na DALL-E, huzalisha picha kutoka kwa maelekezo ya maandishi. Midjourney pia inaweza kutumika kuzalisha picha kwa kutumia maelekezo kama "paka aliyevaa kofia", au "mbwa mwenye mohawk".

*Transformer autoregressive* huamua jinsi mfano unavyotengeneza picha kutoka kwa maelezo ya maandishi, huunda pikseli moja kwa moja, kisha hutumia pikseli zilizotengenezwa kuzalisha pikseli inayofuata. Kupitia tabaka nyingi katika mtandao wa neva, hadi picha kukamilika.

## Kujenga programu yako ya kwanza ya uzalishaji wa picha

Basi inachukua nini kujenga programu ya uzalishaji wa picha? Unahitaji maktaba zifuatazo:

- **python-dotenv**, inashauriwa sana kutumia maktaba hii kuweka siri zako ndani ya faili la *.env* mbali na msimbo.
- **openai**, maktaba hii ndio utakayotumia kuwasiliana na API ya OpenAI.
- **pillow**, kwa ajili ya kufanya kazi na picha katika Python.
- **requests**, ili kusaidia kufanya maombi ya HTTP.


1. Tengeneza faili *.env* yenye maudhui yafuatayo:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. Kusanya maktaba zilizo juu katika faili iitwayo *requirements.txt* kama ifuatavyo:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. Ifuatayo, tengeneza mazingira pepe na sakinisha maktaba:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Kwa Windows, tumia amri zifuatazo kuunda na kuanzisha mazingira yako halisi (virtual environment):

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. Ongeza msimbo ufuatao kwenye faili liitwalo *app.py*:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # ingiza dotenv
    dotenv.load_dotenv()

    # Unda kitu cha OpenAI
    client = openai.OpenAI()


    try:
        # Unda picha kwa kutumia API ya uzalishaji picha
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Weka maandishi ya mwongozo wako hapa
            size='1024x1024',
            n=1
        )
        # Weka saraka kwa picha iliyohifadhiwa
        image_dir = os.path.join(os.curdir, 'images')

        # Ikiwa saraka haipo, iunde
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Anzisha njia ya picha (zingatia aina ya faili inapaswa kuwa png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # Pata picha iliyozalishwa
        print(generation_response)

        image_url = generation_response.data[0].url # toa URL ya picha kutoka kwenye jibu
        generated_image = requests.get(image_url).content  # pakua picha
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Onyesha picha katika mtazamaji wa picha wa kawaida
        image = Image.open(image_path)
        image.show()

    # shika makosa
    except openai.BadRequestError as err:
        print(err)

    ```

Hebu tuelezee msimbo huu:

- Kwanza, tunaingiza maktaba tunazohitaji, ikiwa ni pamoja na maktaba ya OpenAI, maktaba ya dotenv, maktaba ya requests, na maktaba ya Pillow.

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv 
    ```

- Baada ya hapo, tunaunda kielekezi, ambacho kitapata ufunguo wa API kutoka kwa faili lako ``.env``.

    ```python
        # Unda kitu cha OpenAI
        client = openai.OpenAI()
    ```

- Ifuatayo, tunazalisha picha:

    ```python
    # Unda picha kwa kutumia API ya kizazi cha picha
    generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Weka maandishi yako ya maelekezo hapa
        size='1024x1024',
        n=1
    )
    ```

    Msimbo uliotajwa juu unatuma jibu la kitu cha JSON chenye URL ya picha iliyotengenezwa. Tunaweza kutumia URL hiyo kupakua picha na kuihifadhi katika faili.

- Mwishowe, tunafungua picha na kutumia mtazamaji wa picha wa kawaida kuionyesha:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### Maelezo zaidi kuhusu kuzalisha picha

Hebu tuangalie msimbo unaozalisha picha kwa undani zaidi:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Ingiza maandishi yako ya maelekezo hapa
        size='1024x1024',
        n=1
    )
```

- **prompt**, ni maandishi ya mwongozo yanayotumika kuzalisha picha. Katika kesi hii, tunatumia mwongozo "Bunny juu ya farasi, akishikilia lolipop, kwenye uwanja wa mvua ambapo hukua daffodils".
- **size**, ni ukubwa wa picha inayozalishwa. Katika kesi hii, tunazalisha picha yenye ukubwa wa pikseli 1024x1024.
- **n**, ni idadi ya picha zinazozalishwa. Katika kesi hii, tunazalisha picha mbili.

Kuna mambo zaidi unayoweza kufanya na picha ambayo tutayazungumzia katika sehemu inayofuata.

## Uwezo wa ziada wa uzalishaji wa picha

Umeona hadi sasa jinsi tulivyoweza kuzalisha picha kwa kutumia mistari michache ya Python. Hata hivyo, kuna mambo zaidi unayoweza kufanya na picha.

Unaweza pia kufanya mambo yafuatayo:

- **Fanya marekebisho**. Kwa kutoa picha iliyopo pamoja na mask na mwongozo, unaweza kubadilisha picha. Kwa mfano, unaweza kuongeza kitu sehemu fulani ya picha. Fikiria picha yetu ya sungura, unaweza kuongeza kofia kwa sungura. Jinsi ya kufanya hivyo ni kwa kutoa picha, mask (kuelezea sehemu inayobadilishwa) na mwongozo wa maandishi kusema kinachopaswa kufanywa.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    Picha msingi itakuwa na sungura tu lakini picha ya mwisho itakuwa na kofia juu ya sungura.
    
- **Tengeneza mabadiliko**. Wazo ni kwamba unachukua picha iliyopo na kuomba mabadiliko yaweze kuundwa. Kuunda mabadiliko, unatoa picha na mwongozo wa maandishi na msimbo kama huu:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
