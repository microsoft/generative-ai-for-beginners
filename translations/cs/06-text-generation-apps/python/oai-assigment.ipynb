## Co je aplikace pro generování textu?

Obvykle když vytváříte aplikaci, má nějaké rozhraní jako následující:

- Příkazové rozhraní. Konzolové aplikace jsou typické aplikace, kde zadáte příkaz a ten provede úkol. Například `git` je příkazová aplikace.
- Uživatelské rozhraní (UI). Některé aplikace mají grafické uživatelské rozhraní (GUI), kde klikáte na tlačítka, zadáváte text, vybíráte možnosti a další.

### Konzolové a UI aplikace jsou omezené

Porovnejte to s příkazovou aplikací, kde zadáváte příkaz:

- **Je omezená**. Nemůžete zadat jakýkoliv příkaz, pouze ty, které aplikace podporuje.
- **Jazykově specifická**. Některé aplikace podporují mnoho jazyků, ale ve výchozím nastavení je aplikace postavena pro specifický jazyk, i když můžete přidat více jazyků.

### Výhody aplikací pro generování textu

Jak je tedy aplikace pro generování textu jiná?

V aplikaci pro generování textu máte větší flexibilitu, nejste omezeni sadou příkazů nebo specifickým vstupním jazykem. Místo toho můžete použít přirozený jazyk k interakci s aplikací. Další výhodou je, že již komunikujete se zdrojem dat, který byl natrénován na rozsáhlém korpusu informací, zatímco tradiční aplikace může být omezená na to, co je v databázi.

### Co mohu vytvořit s aplikací pro generování textu?

Můžete vytvořit mnoho věcí. Například:

- **Chatbot**. Chatbot odpovídající na otázky o tématech, jako je vaše firma a její produkty, by mohl být skvělou volbou.
- **Pomocník**. Velké jazykové modely jsou skvělé na věci jako shrnutí textu, získávání poznatků z textu, tvorbu textů jako životopisy a další.
- **Asistent kódu**. V závislosti na jazykovém modelu, který používáte, můžete vytvořit asistenta kódu, který vám pomáhá psát kód. Například můžete použít produkty jako GitHub Copilot i ChatGPT k pomoci s psaním kódu.

## Generování textu

Pro generování textu použijte metodu `responses.create`. Tady je příklad:

```python

print(response.output_text)
```

V kódu výše vytváříme objekt odpovědi a předáváme model, který chceme použít, a prompt. Pak tiskneme generovaný text.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Vylepšení přidáním filtru a nákupního seznamu

- **Vytvořit nákupní seznam**. Chceme vytvořit nákupní seznam s ohledem na to, co už doma máme.

    Pro tuto funkci bychom mohli zkusit všechno vyřešit v jednom promptu, nebo to rozdělit do dvou promptů. Zkusme druhý přístup. Navrhujeme přidat další prompt, ale aby to fungovalo, musíme výsledek prvního promptu přidat jako kontext k druhému promptu.

    Najděte místo v kódu, kde se vypisuje výsledek z prvního promptu, a přidejte níže následující kód:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # vytiskni odpověď
    print("Shopping list:")
    print(response.output_text)
    ```

    Všimněte si následujícího:

    - Sestavujeme nový prompt tím, že přidáváme výsledek z prvního promptu do nového promptu:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Vytváříme nový požadavek, ale také zohledňujeme počet tokenů, o kterých jsme žádali v prvním promptu, takže tentokrát říkáme, že `max_output_tokens` je 1200.

- **Slovo o délce tokenů**. Měli bychom zvažovat, kolik tokenů potřebujeme k vygenerování požadovaného textu. Tokeny něco stojí, takže tam, kde je to možné, bychom měli být ekonomičtí s jejich použitím. Například, můžeme prompt formulovat tak, abychom použili méně tokenů?

   Pro změnu počtu použitých tokenů můžete použít parametr `max_output_tokens`. Například, pokud chcete použít 100 tokenů, uděláte takto:

    ```python

    ```

- **Experimentování s teplotou**. Teplota je něco, o čem jsme se zatím nezmínili, ale je to důležitý kontext pro výkon našeho programu. Čím vyšší hodnota teploty, tím náhodnější bude výstup. Naopak čím nižší hodnota teploty, tím předvídatelnější bude výstup. Zvažte, zda chcete mít ve svém výstupu variace, či nikoliv.
