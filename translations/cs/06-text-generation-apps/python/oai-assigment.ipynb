## Co je aplikace pro generování textu?

Obvykle když vytváříte aplikaci, má nějaké rozhraní jako následující:

- Na příkazovém řádku. Konzolové aplikace jsou typické aplikace, kde zadáváte příkaz a provádíte úkol. Například `git` je aplikace založená na příkazech.
- Uživatelské rozhraní (UI). Některé aplikace mají grafické uživatelské rozhraní (GUI), kde klikáte na tlačítka, zadáváte text, vybíráte možnosti a podobně.

### Konzolové a GUI aplikace jsou omezené

Porovnejte to s aplikací založenou na příkazech, kde zadáváte příkaz:

- **Je omezená**. Nemůžete zadat libovolný příkaz, jen ty, které aplikace podporuje.
- **Specifická jazyková omezení**. Některé aplikace podporují mnoho jazyků, ale ve výchozím nastavení je aplikace vytvořena pro konkrétní jazyk, i když můžete přidat další podporu jazyků.

### Výhody aplikací pro generování textu

Jak je tedy aplikace pro generování textu jiná?

V aplikaci pro generování textu máte větší flexibilitu, nejste omezeni sadou příkazů nebo konkrétním vstupním jazykem. Místo toho můžete používat přirozený jazyk pro interakci s aplikací. Další výhodou je, že již pracujete se zdrojem dat, který byl natrénován na rozsáhlém korpusu informací, zatímco tradiční aplikace může být omezená na obsah v databázi.

### Co můžu s aplikací pro generování textu vytvořit?

Existuje mnoho věcí, které můžete vytvořit. Například:

- **Chatbota**. Chatbot odpovídající na otázky o tématech, jako je vaše společnost a její produkty, by mohl být dobrým řešením.
- **Asistenta**. LLM jsou skvělé na věci jako shrnování textu, získávání poznatků z textu, tvorbu textu jako životopisů a dalších.
- **Asistenta pro kódování**. Podle použitého modelu můžete vytvořit asistenta pro kódování, který vám pomáhá psát kód. Například můžete použít produkt jako GitHub Copilot i ChatGPT, které vám pomohou psát kód.

## Generování textu

Způsob, jak generovat text, je použít metodu `responses.create`. Zde je příklad:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

V uvedeném kódu vytvoříme objekt odpovědi a zadáme model, který chceme použít, a prompt. Pak tiskneme generovaný text.

### Odpovědi v chatu

Dosud jste viděli, jak generovat text z jediného promptu. Responses API je také dobře vhodné pro chatboty - poskytnete seznam zpráv jako `input` k vytvoření konverzace. Zde je příklad použití:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

### Vylepšení přidáním filtru a nákupního seznamu

- **Vytvořit nákupní seznam**. Chceme vytvořit nákupní seznam s ohledem na to, co už doma máme.

    Pro tuto funkci bychom mohli zkusit vyřešit vše v jednom promptu, nebo to rozdělit do dvou promptů. Zkusme ten druhý přístup. Navrhujeme přidat další prompt, ale aby to fungovalo, musíme přidat výsledek prvního promptu jako kontext k druhému promptu.

    Najděte část kódu, která vypisuje výsledek z prvního promptu, a přidejte níže následující kód:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # vytiskni odpověď
    print("Shopping list:")
    print(response.output_text)
    ```

    Všimněte si následujícího:

    - Sestavujeme nový prompt přidáním výsledku z prvního promptu do nového promptu:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Vytváříme nový požadavek, ale také bereme v úvahu počet tokenů, které jsme požadovali v prvním promptu, tentokrát říkáme `max_output_tokens` je 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Po spuštění tohoto kódu nyní dostaneme následující výstup:

        ```output

        ```
        
- **Slovo o délce tokenů**. Měli bychom zvážit, kolik tokenů potřebujeme k vygenerování požadovaného textu. Tokeny něco stojí, takže se tam, kde je to možné, snažme být k tokenům ekonomičtí. Například, zda můžeme prompt formulovat tak, abychom použili méně tokenů?

   Pro změnu počtu použitých tokenů lze použít parametr `max_output_tokens`. Například pokud chcete použít 100 tokenů, uděláte to takto:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimentování s teplotou**. Teplota je něco, co jsme zatím nezmínili, ale je to důležitý kontext pro chování našeho programu. Čím vyšší hodnota teploty, tím náhodnější bude výstup. Naopak čím nižší teplota, tím předvídatelnější je výstup. Zvažte, zda chcete mít výstup více či méně různorodý.

   Pro změnu teploty lze použít parametr `temperature`. Například pokud chcete teplotu 0.5, uděláte to takto:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Poznámka, čím blíže 1.0, tím rozmanitější bude výstup.
