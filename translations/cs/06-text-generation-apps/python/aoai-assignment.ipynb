## Co je aplikace pro generování textu?

Obvykle, když vytváříte aplikaci, má nějaké rozhraní jako například:

- Příkazové. Konzolové aplikace jsou typické aplikace, kde zadáváte příkaz a aplikace provede úkol. Například `git` je příkazová aplikace.
- Uživatelské rozhraní (UI). Některé aplikace mají grafické uživatelské rozhraní (GUI), kde klikáte na tlačítka, zadáváte text, vybíráte možnosti a další.

### Konzolové a UI aplikace jsou omezené

Porovnejte to s příkazovou aplikací, kde zadáváte příkaz:

- **Je omezená**. Nemůžete zadat jakýkoli příkaz, pouze ty, které aplikace podporuje.
- **Je specifická pro jazyk**. Některé aplikace podporují mnoho jazyků, ale výchozí aplikace je postavená pro konkrétní jazyk, i když můžete přidat podporu dalších jazyků.

### Výhody aplikací pro generování textu

Jak se tedy aplikace pro generování textu liší?

V aplikaci pro generování textu máte větší flexibilitu, nejste omezeni na sadu příkazů nebo konkrétní vstupní jazyk. Místo toho můžete používat přirozený jazyk k interakci s aplikací. Další výhodou je, že již pracujete s datovým zdrojem, který byl natrénován na obrovském korpusu informací, zatímco tradiční aplikace může být omezena tím, co je v databázi.

### Co mohu s aplikací pro generování textu vytvořit?

Můžete vytvořit mnoho věcí, například:

- **Chatbota**. Chatbot, který odpovídá na otázky o tématech, jako je vaše firma a její produkty, může být dobrým řešením.
- **Pomocníka**. Modely LLM jsou skvělé pro věci jako shrnutí textu, získávání postřehů z textu, generování textu jako životopisy a další.
- **Asistenta pro kód**. V závislosti na modelu jazyka, který použijete, můžete vytvořit asistenta pro kód, který vám pomůže psát kód. Například můžete využít produkty jako GitHub Copilot i ChatGPT k pomoci při psaní kódu.

## Generování textu

Způsob, jak generovat text, je použít metodu `responses.create`. Zde je příklad:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Ve výše uvedeném kódu vytvoříme objekt odpovědi a předáme model, který chceme použít, a prompt. Pak vytiskneme vygenerovaný text.

### Odpovědi v chatu

API odpovědí je vhodné jak pro jednorázové generování textu, tak pro vícesmyčkové chatboty – jednoduše poskytujete více zpráv v seznamu `input` k vytvoření konverzace:

### Improve by adding filter and shopping list
