## Co je to aplikace pro generování textu?

Obvykle, když vytváříte aplikaci, má nějaké uživatelské rozhraní, například:

- Příkazové rozhraní. Konzolové aplikace jsou typické aplikace, kde napíšete příkaz a aplikace vykoná úlohu. Například `git` je aplikace založená na příkazech.
- Uživatelské rozhraní (UI). Některé aplikace mají grafické uživatelské rozhraní (GUI), kde klikáte na tlačítka, zadáváte text, vybíráte možnosti a další.

### Konzolové a GUI aplikace jsou omezené

Porovnejte to s aplikací založenou na příkazech, kde napíšete příkaz:

- **Je omezené**. Nemůžete napsat jakýkoliv příkaz, pouze ty, které aplikace podporuje.
- **Specifické pro jazyk**. Některé aplikace podporují mnoho jazyků, ale výchozí je aplikace vytvořená pro konkrétní jazyk, i když můžete přidat podporu dalších jazyků.

### Výhody aplikací pro generování textu

Jak je tedy aplikace pro generování textu jiná?

V aplikaci pro generování textu máte větší flexibilitu, nejste omezeni sadou příkazů nebo konkrétním vstupním jazykem. Místo toho můžete používat přirozený jazyk k interakci s aplikací. Další výhodou je, že již komunikujete se zdrojem dat, který byl natrénován na obrovském množství informací, zatímco tradiční aplikace může být omezena pouze na data v databázi.

### Co mohu vytvořit pomocí aplikace pro generování textu?

Existuje mnoho věcí, které můžete vytvořit. Například:

- **Chatbot**. Chatbot odpovídající na otázky o tématech jako o vaší společnosti a jejích produktech může být dobrou volbou.
- **Pomocník**. Velké jazykové modely (LLM) jsou skvělé v souhrnes textu, získávání poznatků z textu, produkci textu jako životopisy a další.
- **Asistent kódu**. V závislosti na použitém jazykovém modelu můžete vytvořit asistenta, který vám pomůže psát kód. Například můžete použít produkty jako GitHub Copilot i ChatGPT pro pomoc při psaní kódu.

## Cvičení - vytvořte generátor receptů

Nyní, když jsme prošli scénářem, napišme kód odpovídající demonstrovanému scénáři. Postupujte podle těchto kroků:

1. Použijte existující soubor jako výchozí bod
1. Vytvořte proměnnou `prompt` a upravte ukázkový kód takto:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Skvěle, podívejme se, jak můžeme věci vylepšit. Abychom věci zlepšili, chceme zajistit, aby byl kód flexibilní, takže ingredience a počet receptů lze vylepšovat a měnit.

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

### Vylepšení přidáním filtru a nákupního seznamu

Nyní máme funkční aplikaci schopnou generovat recepty a je flexibilní, protože se spoléhá na vstupy od uživatele, jak z hlediska počtu receptů, tak i použitých ingrediencí.

Abychom ji dále vylepšili, chceme přidat následující:

- **Filtrovat ingredience**. Chceme mít možnost filtrovat ingredience, které nemáme rádi nebo na které jsme alergičtí. Pro tuto změnu můžeme upravit náš stávající prompt a přidat na jeho konec filtrovací podmínku takto:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```
