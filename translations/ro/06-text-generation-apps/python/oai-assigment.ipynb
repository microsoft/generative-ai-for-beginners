## Ce este o aplicație de generare text?

De obicei, când construiești o aplicație are un fel de interfață, cum ar fi următoarea:

- Bazată pe comenzi. Aplicațiile de tip consolă sunt tipice: scrii o comandă și ea execută o sarcină. De exemplu, `git` este o aplicație bazată pe comenzi.
- Interfață cu utilizatorul (UI). Unele aplicații au interfețe grafice (GUI) unde faci click pe butoane, introduci text, selectezi opțiuni și altele.

### Aplicațiile de consolă și UI sunt limitate

Comparați cu o aplicație bazată pe comenzi unde scrii o comandă:

- **Este limitată**. Nu poți scrie orice comandă, ci doar pe cele pe care aplicația le suportă.
- **Limbaj specific**. Unele aplicații suportă multe limbi, dar în mod implicit aplicația este construită pentru un anumit limbaj, chiar dacă poți adăuga suport pentru mai multe limbi.

### Beneficiile aplicațiilor de generare text

Deci, cum este diferită o aplicație de generare text?

Într-o aplicație de generare text, ai mai multă flexibilitate, nu ești limitat la un set de comenzi sau un limbaj de intrare specific. În schimb, poți folosi limbajul natural pentru a interacționa cu aplicația. Un alt beneficiu este că, deoarece deja interacționezi cu o sursă de date antrenată pe un vast corpus de informații, în timp ce o aplicație tradițională poate fi limitată la ceea ce conține o bază de date.

### Ce pot construi cu o aplicație de generare text?

Sunt multe lucruri pe care le poți construi. De exemplu:

- **Un chatbot**. Un chatbot care răspunde la întrebări despre subiecte, cum ar fi compania ta și produsele sale, ar putea fi o alegere potrivită.
- **Asistent**. Modelele LLM sunt excelente pentru lucruri precum sumarizarea textului, obținerea de insight-uri din text, generarea de text precum CV-uri și altele.
- **Asistent de cod**. În funcție de modelul de limbaj folosit, poți construi un asistent de cod care să te ajute să scrii cod. De exemplu, poți folosi un produs precum GitHub Copilot, precum și ChatGPT pentru a te ajuta cu scrierea codului.

## Generează text

Modul de a genera text este să folosești metoda `responses.create`. Iată un exemplu:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

În codul de mai sus, creăm un obiect de răspuns și transmitem modelul pe care vrem să îl folosim și promptul. Apoi afișăm textul generat.

### Răspunsuri chat

Până acum, ai văzut cum putem genera text dintr-un singur prompt. API-ul de Răspunsuri este de asemenea potrivit pentru chatbots - oferi o listă de mesaje ca `input` pentru a construi o conversație. Iată un exemplu de folosire a acestuia:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Mai multe despre această funcționalitate într-un capitol viitor.

   Pentru a schimba numărul de tokeni folosiți, puteți folosi parametrul `max_output_tokens`. De exemplu, dacă doriți să folosiți 100 tokeni, ați face astfel:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimentarea cu temperatura**. Temperatura este ceva ce nu am menționat până acum, dar este un context important pentru cum se comportă programul nostru. Cu cât valoarea temperaturii este mai mare, cu atât ieșirea este mai aleatorie. În schimb, cu cât temperatura este mai mică, cu atât ieșirea este mai predictibilă. Gândiți-vă dacă doriți variație în ieșire sau nu.

   Pentru a modifica temperatura, puteți folosi parametrul `temperature`. De exemplu, dacă doriți o temperatură de 0.5, ați face astfel:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Notă, cu cât este mai aproape de 1.0, cu atât ieșirea este mai variată.
