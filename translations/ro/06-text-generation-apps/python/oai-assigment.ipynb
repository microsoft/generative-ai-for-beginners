## Ce este o aplicație de generare text?

De obicei, când construiești o aplicație, aceasta are un fel de interfață, cum ar fi următoarele:

- Bazată pe comenzi. Aplicațiile de consolă sunt aplicații obișnuite în care tastezi o comandă iar aceasta realizează o sarcină. De exemplu, `git` este o aplicație bazată pe comenzi.
- Interfață cu utilizatorul (UI). Unele aplicații au interfețe grafice (GUI) unde apeși butoane, introduci text, selectezi opțiuni și altele.

### Aplicațiile de consolă și UI sunt limitate

Compară asta cu o aplicație bazată pe comenzi unde tastezi o comandă:

- **Este limitată**. Nu poți introduce orice comandă, doar pe cele pe care le susține aplicația.
- **Este specifică limbajului**. Unele aplicații susțin mai multe limbi, dar implicit aplicația este construită pentru un limbaj specific, chiar dacă poți adăuga suport pentru mai multe limbi.

### Beneficiile aplicațiilor de generare text

Atunci, cum este diferită o aplicație de generare text?

Într-o aplicație de generare text, ai mai multă flexibilitate, nu ești limitat la un set anume de comenzi sau un limbaj de intrare specific. În schimb, poți folosi limbajul natural pentru a interacționa cu aplicația. Un alt beneficiu este că deja interacționezi cu o sursă de date care a fost antrenată pe un vast corpus de informații, pe când o aplicație tradițională poate fi limitată la ceea ce este în baza de date.

### Ce pot construi cu o aplicație de generare text?

Sunt multe lucruri pe care le poți construi. De exemplu:

- **Un chatbot**. Un chatbot care răspunde la întrebări despre subiecte, precum compania ta și produsele sale, ar putea fi o alegere bună.
- **Asistent**. LLM-urile sunt excelente la lucruri precum sumarizarea textului, obținerea de insight-uri din text, generarea de texte precum CV-uri și altele.
- **Asistent de cod**. În funcție de modelul de limbaj pe care îl folosești, poți construi un asistent de cod care te ajută să scrii cod. De exemplu, poți folosi un produs ca GitHub Copilot sau ChatGPT pentru a te ajuta să scrii cod.

## Generarea textului

Modul de a genera text este folosind metoda `responses.create`. Iată un exemplu:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

În codul de mai sus, creăm un obiect de răspuns și îi trimitem modelul pe care vrem să îl folosim și promptul. Apoi afișăm textul generat.

### Răspunsuri de chat

Până acum, ai văzut cum putem genera text dintr-un singur prompt. API-ul Responses este de asemenea potrivit pentru chatbots - oferi o listă de mesaje ca `input` pentru a construi o conversație. Iată un exemplu de folosire:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Mai multe despre această funcționalitate într-un capitol viitor.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

   Pentru a schimba numărul de tokeni folosiți, poți folosi parametrul `max_output_tokens`. De exemplu, dacă vrei să folosești 100 de tokeni, ai face:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimentarea cu temperatura**. Temperatura este ceva ce nu am menționat până acum, dar este un context important pentru cum performează programul nostru. Cu cât valoarea temperaturii este mai mare, cu atât rezultatul este mai aleatoriu. Invers, cu cât temperatura este mai scăzută, cu atât rezultatul este mai predictibil. Gândește-te dacă dorești variație în output sau nu.

   Pentru a modifica temperatura, poți folosi parametrul `temperature`. De exemplu, dacă vrei să folosești temperatura de 0.5, ai face:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Atenție, cu cât ne apropiem de 1.0, cu atât output-ul este mai variat.
