## מהי אפליקציית יצירת טקסט?

בדרך כלל כשבונים אפליקציה יש לה איזשהו ממשק כמו הבא:

- מבוסס פקודות. אפליקציות קונסול הן אפליקציות טיפוסיות שבהן אתם מקלידים פקודה והיא מבצעת משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). באפליקציות מסוימות יש ממשקי משתמש גרפיים (GUIs) שבהם לוחצים על כפתורים, מזינים טקסט, בוחרים אפשרויות ועוד.

### אפליקציות קונסול ו-UI מוגבלות

השוו זאת לאפליקציית מבוססת פקודה שבה מקלידים פקודה: 

- **הן מוגבלות**. אי אפשר פשוט להקליד כל פקודה, רק את אלה שהאפליקציה תומכת בהם.
- **תלויות בשפה ספציפית**. אפליקציות מסוימות תומכות בשפות רבות, אך כברירת מחדל האפליקציה בנויה לשפה מסוימת, גם אם ניתן להוסיף תמיכה בשפות נוספות.

### יתרונות של אפליקציות יצירת טקסט

אז איך אפליקציית יצירת טקסט שונה?

באפליקציית יצירת טקסט יש יותר גמישות, אתם לא מוגבלים למערך של פקודות או שפת קלט ספציפית. במקום זאת, ניתן להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שכבר מתקשרים עם מקור נתונים שאומן על אוסף עצום של מידע, בעוד שאפליקציה מסורתית עלולה להיות מוגבלת למה שיש בבסיס נתונים.

### מה אני יכול לבנות עם אפליקציית יצירת טקסט?

יש הרבה דברים שניתן לבנות. לדוגמה:

- **צ'אטבוט**. צ'אטבוט שעונה על שאלות בנושאים כמו החברה שלך והמוצרים שלה יכול להתאים מאד.
- **עוזר**. LLMs מצוינים בדברים כמו סיכום טקסט, הפקת תובנות מטקסט, יצירת טקסט כמו קורות חיים ועוד.
- **עוזר לקוד**. בהתאם לדגם השפה שבו משתמשים, ניתן לבנות עוזר קוד שעוזר לכתוב קוד. לדוגמה, ניתן להשתמש במוצר כמו GitHub Copilot וכן ב-ChatGPT כדי לעזור בכתיבת קוד.

## תרגיל - בנה מחולל מתכונים

עכשיו כששיחקנו תרחיש, בואו נכתוב קוד שיתאים לתרחיש שהוצג. לשם כך, בצעו את השלבים הבאים:

1. השתמשו בקובץ הקיים כנקודת התחלה
1. צרו משתנה `prompt` ושנו את קוד הדוגמה כך:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

מצוין, נבדוק כיצד ניתן לשפר את הדברים. כדי לשפר את הדברים, אנו רוצים לוודא שהקוד גמיש, כך שניתן לשפר ולשנות את החומרים ומספר המתכונים.


1. בואו נשנה את הקוד באופן הבא:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
