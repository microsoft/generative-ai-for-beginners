## מהי אפליקציית יצירת טקסט?

בדרך כלל כשאתם בונים אפליקציה יש לה ממשק מסוג כלשהו כמו הבא:

- מבוססת פקודות. אפליקציות קונסול הן אפליקציות טיפוסיות שבהן אתם מקלידים פקודה והן מבצעות משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). באפליקציות מסוימות יש ממשקי משתמש גרפיים (GUIs) שבהם אתם לוחצים על כפתורים, מזינים טקסט, בוחרים אפשרויות ועוד.

### אפליקציות קונסול ו-UI מוגבלות

השוו זאת לאפליקציה מבוססת פקודות בה אתם מקלידים פקודה:

- **זה מוגבל**. אתם לא יכולים להקליד כל פקודה, רק את אלו שהאפליקציה תומכת בהן.
- **מסוגל לשפה ספציפית**. יש אפליקציות שתומכות בשפות רבות, אבל כברירת מחדל האפליקציה מותאמת לשפה מסוימת, אפילו אם ניתן להוסיף תמיכה בשפות נוספות.

### יתרונות אפליקציות ליצירת טקסט

אז איך אפליקציית יצירת טקסט שונה?

באפליקציית יצירת טקסט יש לכם יותר גמישות, אינכם מוגבלים לסט פקודות או שפת קלט ספציפית. במקום זאת, תוכלו להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שכיוון שאתם כבר מתקשרים עם מקור מידע שעבר אימון על מאגר מידע עצום, בניגוד לאפליקציה מסורתית שיכולה להיות מוגבלת במה שיש במסד נתונים.

### מה ניתן לבנות עם אפליקציית יצירת טקסט?

ישנן המון אפשרויות לבנות. לדוגמה:

- **צ'אטבוט**. צ'אטבוט שעונה על שאלות בנושאים שונים, כמו החברה שלכם והמוצרים שלה, יכול להתאים היטב.
- **עוזר**. LLMs מצטיינים בדברים כמו סיכום טקסט, קבלת תובנות מטקסט, יצירת טקסט כמו קורות חיים ועוד.
- **עוזר קוד**. בהתאם לדגם השפה שבו אתם משתמשים, תוכלו לבנות עוזר קוד שעוזר לכם לכתוב קוד. לדוגמה, אתם יכולים להשתמש במוצר כמו GitHub Copilot וגם ב-ChatGPT כדי לעזור לכם לכתוב קוד.

## תרגיל - לבנות גנרטור למתכונים

עכשיו כששיחקנו תרחיש, בואו נכתוב קוד שתואם לתרחיש שהודגם. לשם כך, עקבו אחרי השלבים הבאים:

1. השתמשו בקובץ הקיים כנקודת התחלה
1. צרו משתנה `prompt` ושנו את קוד הדוגמה כך:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. בואו נשנה את הקוד בדרך הבאה:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
