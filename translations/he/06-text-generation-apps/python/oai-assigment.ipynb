## מהי אפליקציית יצירת טקסט?

בדרך כלל כשבונים אפליקציה יש לה ממשק כלשהו כמו הבא:

- מבוססת פקודות. אפליקציות קונסול הן אפליקציות טיפוסיות בהן אתה מקליד פקודה והיא מבצעת משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). יש אפליקציות עם ממשקי משתמש גרפיים (GUIs) שבהם אתה לוחץ על כפתורים, מזין טקסט, בוחר אפשרויות ועוד.

### אפליקציות קונסול וממשקי משתמש מוגבלות

השווה זאת לאפליקציה מבוססת פקודה שבה אתה מקליד פקודה:

- **זו מגבלה**. אי אפשר להקליד כל פקודה שרוצים, רק את אלה שהאפליקציה תומכת בהם.
- **ספציפית לשפה**. יש אפליקציות התומכות בשפות רבות, אך כברירת מחדל האפליקציה בנויה לשפה ספציפית, אפילו שאפשר להוסיף תמיכה לשפות נוספות.

### יתרונות של אפליקציות יצירת טקסט

אז איך אפליקציית יצירת טקסט שונה?

באפליקציית יצירת טקסט יש לך יותר גמישות, אינך מוגבל לקבוצת פקודות או שפת קלט ספציפית. במקום זאת, ניתן להשתמש בשפה טבעית כדי לקיים אינטראקציה עם האפליקציה. יתרון נוסף הוא שכיוון שאתה כבר מתקשר עם מקור מידע שאומן על קורפוס רחב של מידע, בעוד שאפליקציה מסורתית עשויה להיות מוגבלת למה שיש במאגר נתונים.

### מה אפשר לבנות עם אפליקציית יצירת טקסט?

יש הרבה דברים שאפשר לבנות. לדוגמה:

- **צ'אטבוט**. צ'אטבוט שמענה על שאלות בנושאים, כמו החברה שלך ומוצריה, יכול להתאים מאוד.
- **עוזר**. מודלים גדולים של שפה מעולים בדברים כמו סיכום טקסט, חילוץ תובנות מהטקסט, יצירת טקסטים כמו קורות חיים ועוד.
- **עוזר קוד**. תלוי במודל השפה בו אתה משתמש, תוכל לבנות עוזר קוד שמסייע לך לכתוב קוד. לדוגמה, ניתן להשתמש במוצר כמו GitHub Copilot וגם ב-ChatGPT כדי לעזור בכתיבת קוד.

## יצירת טקסט

הדרך לייצר טקסט היא להשתמש בשיטה `responses.create`. הנה דוגמה:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

בקוד שלמעלה, אנו יוצרים אובייקט תגובה ומעבירים את המודל שאנו רוצים להשתמש בו ואת הפרומפט. לאחר מכן מדפיסים את הטקסט שנוצר.

### תגובות צ'אט

עד כה ראית איך ליצור טקסט מפרומפט יחיד. ה-Responses API מתאים גם לצ'אטבוטים - מספקים רשימת הודעות כ-`input` כדי לבנות שיחה. הנה דוגמה לשימוש בזה:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

עוד על הפונקציונליות הזו בפרק הבא.

1. צור סביבת פיתוח וירטואלית והתקן את openai:

## תרגיל - לבנות מחולל מתכונים

עכשיו כששיחקנו תרחיש, בואו נכתוב קוד שמתאים לתרחיש שהודגם. כדי לעשות זאת, בצעו את השלבים הבאים:

1. השתמשו בקובץ *app.py* הקיים כנקודת התחלה
1. מצאו את המשתנה `prompt` ושנו את הקוד שלו כדלהלן:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

מצוין, בוא נראה כיצד נוכל לשפר את הדברים. כדי לשפר את הדברים, אנו רוצים לוודא שהקוד גמיש, כך שניתן יהיה לשפר ולשנות את המרכיבים ומספר המתכונים.


1. בואו נשנה את הקוד בדרך הבאה:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

לקיחת הקוד להרצה ניסיונית, יכולה להיראות כך:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

### שיפור על ידי הוספת פילטר ורשימת קניות

כעת יש לנו אפליקציה עובדת המסוגלת להפיק מתכונים והיא גמישה כיוון שהיא מתבססת על קלטים מהמשתמש, הן במספר המתכונים והן במרכיבים בהם משתמשים.

כדי לשפר זאת עוד יותר, אנו רוצים להוסיף את הדברים הבאים:

- **מסננים מרכיבים**. אנחנו רוצים להיות מסוגלים לסנן מרכיבים שאנחנו לא אוהבים או אלרגיים אליהם. כדי לבצע שינוי זה, נוכל לערוך את הפקודה הקיימת ולהוסיף תנאי סינון בסופה כך:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

- **יצירת רשימת קניות**. אנו רוצים ליצור רשימת קניות בהתחשב במה שכבר יש לנו בבית.

    עבור פונקציונליות זו, נוכל לנסות לפתור הכול בפקודה אחת או לחלק אותה לשתי פקודות. ננסה את הגישה השנייה. כאן אנו מציעים להוסיף פקודה נוספת, אך כדי שזה יעבוד, נצטרך להוסיף את תוצאת הפקודה הקודמת כקונטקסט לפקודה השנייה.

    אתר את החלק בקוד שמדפיס את התוצאה מהפקודה הראשונה והוסף את הקוד הבא למטה:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # הדפס תגובה
    print("Shopping list:")
    print(response.output_text)
    ```

    שים לב לדברים הבאים:

    - אנחנו בונים פקודה חדשה על ידי הוספת התוצאה מהפקודה הראשונה לפקודה החדשה:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - אנו מבצעים בקשה חדשה, אך מתחשבים גם במספר הטוקנים שביקשנו בפעם הראשונה, ולכן הפעם נאמר ש-`max_output_tokens` הוא 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        כשרץ את הקוד הזה, אנו מגיעים כעת לפלט הבא:

- **מילה על אורך הטוקנים**. כדאי לשקול כמה טוקנים אנחנו צריכים כדי לייצר את הטקסט הרצוי. טוקנים עולים כסף, לכן כשאפשר, כדאי לנסות להיות כלכליים במספר הטוקנים שבהם משתמשים. למשל, האם ניתן לנסח את הפקודה כך שנשתמש בפחות טוקנים?

   כדי לשנות את מספר הטוקנים, ניתן להשתמש בפרמטר `max_output_tokens`. לדוגמה, אם רוצים להשתמש ב-100 טוקנים, עושים כך:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **ניסוי עם טמפרטורה**. טמפרטורה היא משהו שעד עכשיו לא הזכרנו אך היא חשובה להקשר של איך התוכנית שלנו מתפקדת. ככל שערך הטמפרטורה גבוה יותר, הפלט יהיה אקראי יותר. להיפך, ככל שערך הטמפרטורה נמוך יותר, הפלט יהיה צפוי יותר. שקול אם אתה רוצה גיוון בפלט שלך או לא.

   כדי לשנות את הטמפרטורה, ניתן להשתמש בפרמטר `temperature`. לדוגמה, אם רוצים להשתמש בטמפרטורה של 0.5, עושים כך:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > שים לב, ככל שהערך קרוב יותר ל-1.0, הפלט יהיה מגוון יותר.
