## מהי אפליקציה ליצירת טקסט?

בדרך כלל כשאתה בונה אפליקציה, יש לה ממשק כלשהו, לדוגמה:

- מבוסס פקודות. אפליקציות קונסול הן דוגמה קלאסית, בהן אתה מקליד פקודה והאפליקציה מבצעת משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). יש אפליקציות עם ממשק גרפי (GUI) שבו אתה לוחץ על כפתורים, מזין טקסט, בוחר אפשרויות ועוד.

### אפליקציות קונסול ו-UI מוגבלות

השווה זאת לאפליקציה מבוססת פקודות שבה אתה מקליד פקודה:

- **זה מוגבל**. אי אפשר להקליד כל פקודה שרוצים, רק את אלו שהאפליקציה תומכת בהן.
- **תלוי שפה**. יש אפליקציות שתומכות במספר שפות, אבל לרוב הן נבנות לשפה מסוימת, גם אם אפשר להוסיף תמיכה בשפות נוספות.

### יתרונות של אפליקציות ליצירת טקסט

אז במה שונה אפליקציה ליצירת טקסט?

באפליקציה כזו יש לך הרבה יותר גמישות, אתה לא מוגבל לסט פקודות מסוים או לשפת קלט מסוימת. במקום זאת, אפשר להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שאתה כבר עובד מול מקור מידע שאומן על כמות עצומה של מידע, בעוד שאפליקציה מסורתית מוגבלת למה שיש לה בבסיס הנתונים.

### מה אפשר לבנות עם אפליקציה ליצירת טקסט?

יש הרבה דברים שאפשר לבנות. לדוגמה:

- **צ'אט-בוט**. צ'אט-בוט שעונה על שאלות בנושאים שונים, כמו החברה שלך והמוצרים שלה, יכול להתאים מאוד.
- **עוזר**. מודלים גדולים של שפה מצוינים בדברים כמו סיכום טקסטים, הפקת תובנות מטקסט, יצירת טקסטים כמו קורות חיים ועוד.
- **עוזר קוד**. תלוי במודל השפה שבו תשתמש, אפשר לבנות עוזר קוד שיעזור לך לכתוב קוד. לדוגמה, אפשר להשתמש במוצרים כמו GitHub Copilot וגם ב-ChatGPT כדי לעזור בכתיבת קוד.

## יצירת טקסט

הדרך ליצור טקסט היא להשתמש במחלקה `chat.completion`. הנה דוגמה:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

בקוד למעלה, אנחנו יוצרים אובייקט completion ומעבירים לו את המודל שבו נרצה להשתמש ואת הפרומפט. לאחר מכן מדפיסים את הטקסט שנוצר.

## תרגול - האפליקציה הראשונה שלך ליצירת טקסט

עכשיו, אחרי שלמדנו איך להגדיר ולחבר את שירות OpenAI, הגיע הזמן לבנות את אפליקציית יצירת הטקסט הראשונה שלך. כדי לבנות את האפליקציה, בצע את השלבים הבאים:


1. צרו סביבה וירטואלית והתקינו את openai:

  > [!NOTE] שלב זה אינו נחוץ אם אתם מריצים את המחברת הזו ב-Codespaces או בתוך Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> אם אתה משתמש ב-Windows, הקלד `venv\Scripts\activate` במקום `source venv/bin/activate`.

> [!NOTE]
> מצא את המפתח שלך ל-OpenAI על ידי כניסה לכתובת https://platform.openai.com/settings/organization/api-keys וחיפוש אחר `API keys`. תוכל ליצור שם מפתח חדש ולהעתיק מיד את הערך.


1. צור קובץ בשם *app.py* וכתוב בו את הקוד הבא:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output

    ```

## תרגיל - לבנות מחולל מתכונים

עכשיו אחרי ששיחקנו את התרחיש, בואו נכתוב קוד שיתאים למה שהדגמנו. כדי לעשות זאת, בצעו את השלבים הבאים:

1. השתמשו בקובץ *app.py* הקיים כנקודת התחלה
1. מצאו את המשתנה `prompt` ושנו את הקוד שלו לקוד הבא:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

מעולה, בוא נראה איך אפשר לשפר את הדברים. כדי לשפר, נרצה לוודא שהקוד גמיש, כך שאפשר יהיה לשפר ולשנות את המרכיבים ואת מספר המתכונים.


1. בואו נשנה את הקוד בדרך הבאה:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

הרצת הקוד לבדיקה יכולה להיראות כך:

```output
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be.
