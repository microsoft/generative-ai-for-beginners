## מהי אפליקציית יצירת טקסט?  

בדרך כלל כשבונים אפליקציה יש לה ממשק ממש כמו הבא:  

- מבוססת פקודות. אפליקציות קונסולה הן אפליקציות טיפוסיות שבהן אתם מקלידים פקודה והן מבצעות משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודה.  
- ממשק משתמש (UI). יש אפליקציות עם ממשקי משתמש גרפיים (GUIs) שבהם אתם לוחצים על כפתורים, מזינים טקסט, בוחרים אפשרויות ועוד.  

### אפליקציות קונסולה וממשק משתמש מוגבלות  

השוו זאת לאפליקציה מבוססת פקודות שבה אתם מקלידים פקודה:  

- **זו מוגבלת**. אתם לא יכולים להקליד כל פקודה, רק את אלו שהאפליקציה תומכת בהן.  
- **ספציפית לשפה**. יש אפליקציות שתומכות בשפות רבות, אך כברירת מחדל האפליקציה בנויה לשפה מסוימת, אפילו אם אפשר להוסיף תמיכה בשפות נוספות.  

### יתרונות של אפליקציות יצירת טקסט  

אז במה אפליקציית יצירת טקסט שונה?  

באפליקציית יצירת טקסט יש לכם יותר גמישות, אתם לא מוגבלים למערך פקודות או לשפת קלט מסוימת. במקום זאת, אפשר להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שכאשר אתם כבר מתקשרים עם מקור מידע שעבר אימון על מאגר עצום של מידע, בעוד שאפליקציה מסורתית עשויה להיות מוגבלת למה שיש בבסיס נתונים.  

### מה אפשר לבנות עם אפליקציית יצירת טקסט?  

יש רבות שאפשר לבנות. למשל:  

- **צ’אטבוט**. צ’אטבוט שעונה על שאלות בנושא, כמו על החברה ומוצריה יכול להיות התאמה טובה.  
- **עוזר**. מודלים לשוניים גדולים מצטיינים בדברים כמו סיכום טקסט, הוצאת תובנות מהטקסט, יצירת טקסטים כמו קורות חיים ועוד.  
- **עוזר לתכנות**. תלוי במודל השפה שאתם משתמשים, אפשר לבנות עוזר כתיבת קוד. לדוגמה, אפשר להשתמש במוצר כמו GitHub Copilot וכן ChatGPT כדי לעזור לכם לכתוב קוד.

### הגדרת תצורת OpenAI  

אם אתם משתמשים ב-OpenAI, כך מגדירים את התצורה:  

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```
  
כאן אנחנו מגדירים את הדברים הבאים:  

- `api_key`, זה מפתח ה-API שנמצא בלוח הבקרה של OpenAI.  
- `deployment`, זוהי גרסת GPT שלכם.  

> [!NOTE]  
> `os.environ` היא פונקציה שקוראת משתני סביבה. אפשר להשתמש בה כדי לקרוא משתני סביבה כמו `OPENAI_API_KEY`.  

## יצירת טקסט  

הדרך ליצור טקסט היא להשתמש בשיטת `responses.create`. הנה דוגמה:  

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```
  
בקוד שלמעלה, אנו יוצרים אובייקט תגובה ומעבירים את המודל הרצוי ואת המונח (prompt). אז מדפיסים את הטקסט שנוצר.  

### תגובות בצ’אט  

עד כה ראיתם איך אפשר ליצור טקסט ממונח יחיד. ממשק ה-Responses מתאים מאוד גם לצ’אטבוטים – אתם מספקים רשימת הודעות כ-`input` כדי לבנות שיחה. הנה דוגמה לשימוש בזה:  

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```
  
עוד על הפונקציונליות הזו בפרק שיבוא.  

## תרגיל - אפליקציית יצירת הטקסט הראשונה שלכם  

כעת שלמדנו להגדיר ולכונן את שירות OpenAI, הגיע הזמן לבנות את אפליקציית יצירת הטקסט הראשונה שלכם. כדי לבנות את האפליקציה, עקבו אחרי השלבים הבאים:  


1. צור סביבה וירטואלית והתקן את openai:

  > [!NOTE] שלב זה אינו הכרחי אם אתה מפעיל פתק זה ב-Codespaces או בתוך Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

## תרגיל - בניית מחולל מתכונים

עכשיו שכבר שיחקנו תרחיש, בואו נכתוב קוד שמתאים לתרחיש שהודגם. לשם כך, עקבו אחרי הצעדים הבאים:

1. השתמשו בקובץ *app.py* הקיים כנקודת התחלה
1. אתר את המשתנה `prompt` ושנה את הקוד שבו לקוד הבא:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

מצויין, בוא נראה איך אפשר לשפר דברים. כדי לשפר דברים, אנו רוצים לוודא שהקוד גמיש, כך שניתן יהיה לשפר ולשנות את המצרכים ומספר המתכונים.


1. בוא נשנה את הקוד בדרך הבאה:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

לקיחת הקוד להרצת בדיקה, יכולה להראות כך:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

### שיפור על ידי הוספת סינון ורשימת קניות

עכשיו יש לנו אפליקציה עובדת המסוגלת להפיק מתכונים והיא גמישה מכיוון שהיא מסתמכת על קלט מהמשתמש, הן במספר המתכונים והן במרכיבים שבהם משתמשים.

כדי לשפר אותה עוד יותר, אנו רוצים להוסיף את הדברים הבאים:

- **לסנן מרכיבים**. אנו רוצים להיות מסוגלים לסנן מרכיבים שאיננו אוהבים או אלרגיים אליהם. כדי לבצע שינוי זה, נוכל לערוך את הפרומפט הקיים ולהוסיף תנאי סינון בסופו כך:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    למעלה, אנו מוסיפים `{filter}` לסוף הפרומפט וגם תופסים את ערך הסינון מהמשתמש.

    דוגמת קלט להרצת התכנית יכולה להיראות כך עכשיו:

    כפי שניתן לראות, כל מתכון המכיל חלב סונן החוצה. אך, אם אתה רגיש ללקטוז, ייתכן שתרצה לסנן מתכונים המכילים גבינה גם כן, לכן יש צורך להיות ברור.

- **להפיק רשימת קניות**. אנו רוצים להפיק רשימת קניות, בהתחשב במה שכבר יש לנו בבית.

    עבור פונקציונליות זו, נוכל לנסות לפתור את הכל בפרומפט אחד או לחלק את זה לשני פרומפטים. בוא ננסה את הגישה השנייה. כאן אנו מציעים להוסיף פרומפט נוסף, אך על מנת שזה יעבוד, עלינו להוסיף את התוצאה של הפרומפט הראשון כקונטקסט לפרומפט השני.

    אתר את החלק בקוד שמדפיס את התוצאה מהפרומפט הראשון והוסף את הקוד הבא למטה:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # הדפס תגובה
    print("Shopping list:")
    print(response.output_text)
    ```

    שים לב לדברים הבאים:

    - אנו בונים פרומפט חדש על ידי הוספת התוצאה מהפרומפט הראשון לפרומפט החדש:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - אנו מבצעים בקשה חדשה, אך גם מתחשבים במספר הטוקנים שביקשנו בפרומפט הראשון, לכן הפעם אנו אומרים ש`max_output_tokens` הוא 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        לקיחת הקוד הזה לסיבוב, אנו כעת מגיעים לפלט הבא:

- **מילה על אורך טוקנים**. עלינו לשקול כמה טוקנים נצטרך לייצר את הטקסט הרצוי. טוקנים עולים כסף, אז כאשר אפשר, כדאי לנסות להיות חיסכוניים עם מספר הטוקנים בהם אנו משתמשים. לדוגמה, האם ניתן לנסח את הפרומפט כך שנוכל להשתמש בטוקנים פחותים?

   לשינוי מספר הטוקנים, ניתן להשתמש בפרמטר `max_output_tokens`. לדוגמה, אם רוצים להשתמש ב-100 טוקנים, תעשו כך:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **ניסויים עם טמפרטורה**. טמפרטורה היא משהו שעד כה לא הזכרנו אך הוא בהיבט חשוב של איך התכנית שלנו מתפקדת. ככל שערך הטמפרטורה גבוה יותר, הפלט יהיה אקראי יותר. לעומת זאת, ככל שערך הטמפרטורה נמוך יותר, הפלט יהיה צפוי יותר. שקלו אם אתם רוצים שונות בפלט או לא.

   לשינוי הטמפרטורה, ניתן להשתמש בפרמטר `temperature`. לדוגמה, אם רוצים להשתמש בטמפרטורה של 0.5, תעשו כך:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > שים לב, ככל שקרוב יותר ל-1.0, הפלט יהיה מגוון יותר.
