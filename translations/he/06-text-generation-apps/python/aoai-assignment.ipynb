## מהי אפליקציית יצירת טקסט?

בדרך כלל כשאתם בונים אפליקציה יש לה סוג של ממשק כמו הבא:

- מבוסס פקודות. אפליקציות קונסולה הן אפליקציות טיפוסיות שבהן אתם מקלידים פקודה והיא מבצעת משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). יש אפליקציות עם ממשק משתמש גרפי (GUI) שבהן אתם לוחצים על כפתורים, מזינים טקסט, בוחרים אפשרויות ועוד.

### אפליקציות קונסולה וממשק משתמש מוגבלות

השוו זאת לאפליקציה מבוססת פקודה שבה אתם מקלידים פקודה:

- **הן מוגבלות**. לא תוכלו להקליד כל פקודה, אלא רק את הפקודות שהאפליקציה תומכת בהן.
- **שפתיות ספציפית**. יש אפליקציות התומכות בשפות רבות, אך כברירת מחדל האפליקציה בנויה עבור שפה מסוימת, גם אם ניתן להוסיף תמיכה בשפות נוספות.

### יתרונות אפליקציות יצירת טקסט

אז מה שונה באפליקציית יצירת טקסט?

באפליקציית יצירת טקסט יש לכם יותר גמישות, אתם לא מוגבלים למערך פקודות או שפת קלט מסוימת. תוכלו להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שברגע שאתם מתקשרים עם מקור מידע שהוכשר על קורפוס עצום של מידע, בעוד שאפליקציה מסורתית עשויה להיות מוגבלת למה שיש בבסיס הנתונים.

### מה אני יכול לבנות עם אפליקציית יצירת טקסט?

יש הרבה דברים שאפשר לבנות. לדוגמה:

- **צ'אטבוט**. צ'אטבוט שמענה על שאלות בנושאים, כמו החברה שלכם ומוצריה, יכול להתאים.
- **עוזר**. מודלים גדולים של שפה מצטיינים בדברים כמו סיכום טקסט, הפקת תובנות מטקסט, יצירת טקסט כמו קורות חיים ועוד.
- **עוזר קוד**. בהתאם למודל השפה שבו אתם משתמשים, תוכלו לבנות עוזר קוד שיעזור לכם לכתוב קוד. לדוגמה, תוכלו להשתמש במוצרים כמו GitHub Copilot וגם ChatGPT שיעזרו לכם לכתוב קוד.

### מיקום מפתח ה-API והנקודת סיום

בשלב זה, עליכם להגדיר לספריית `openai` איזה מפתח API להשתמש בו. כדי למצוא את מפתח ה-API שלכם, גשו למדור "Keys and Endpoint" במשאב Azure OpenAI והעתיקו את הערך "Key 1".

עכשיו שאתם מצוידים במידע זה, ננחה את הספריות להשתמש בו.

> [!NOTE]
> כדאי להפריד את מפתח ה-API שלכם מהקוד. אפשר לעשות זאת על ידי שימוש במשתני סביבה.
> - הגדירו את משתנה הסביבה `AZURE_OPENAI_API_KEY` למפתח ה-API שלכם בקובץ .env. אם כבר השלמתם את התרגילים הקודמים בקורס זה, אתם מוכנים.


### הגדרת תצורה של Azure

אם אתם משתמשים ב-Azure OpenAI, כך תגדירו תצורה. ממשק התגובות API מופעל מנקודת הסיום של Azure OpenAI **v1**, לכן נפנה את הלקוח `OpenAI` ל-`<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

למעלה אנו מגדירים את הדברים הבאים:

- `api_key`, זהו מפתח ה-API שלכם שנמצא ב-Azure Portal.
- `base_url`, זו נקודת הסיום של Azure OpenAI שלכם עם התוספת `/openai/v1/`. ניתן למצוא את נקודת הסיום ב-Azure Portal לצד מפתח ה-API שלכם. שימוש בנקודת הסיום v1 אומר שאין צורך להעביר `api_version`.
- `deployment`, זהו שם הפריסה של המודל שיצרתם בפורטל Foundry.

> [!NOTE]
> `os.environ` היא מפתחות ששואבת משתני סביבה. אפשר להשתמש בה כדי לקרוא משתני סביבה כמו `AZURE_OPENAI_API_KEY` ו-`AZURE_OPENAI_ENDPOINT`.

## יצירת טקסט

הדרך ליצור טקסט היא באמצעות השיטה `responses.create`. הנה דוגמה:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

בקוד למעלה יצרנו אובייקט תגובה והעברנו את המודל שאנו רוצים להשתמש בו ואת הפרומפט. לאחר מכן הדפסנו את הטקסט שנוצר.

1. צור סביבה וירטואלית והתקן את openai:

## תרגיל - לבנות יוצר מתכונים

עכשיו כששיחקנו תרחיש, בואו נכתוב קוד שתואם לתרחיש שהוצג. לשם כך, בצעו את השלבים הבאים:

1. השתמשו בקובץ *app.py* הקיים כנקודת התחלה
1. מצאו את המשתנה `prompt` ושנו את הקוד שלו למה שבא אחרי:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
