## מהי אפליקציית יצירת טקסט?

בדרך כלל, כשאתם בונים אפליקציה יש לה ממשק מסוים כמו הבא:

- מבוססת פקודות. אפליקציות קונסולה הן אפליקציות טיפוסיות שבהן אתם מקלידים פקודה והיא מבצעת משימה. לדוגמה, `git` היא אפליקציה מבוססת פקודות.
- ממשק משתמש (UI). יש אפליקציות עם ממשק משתמש גרפי (GUI) שבו אתם לוחצים על כפתורים, מקלידים טקסט, בוחרים אפשרויות ועוד.

### אפליקציות קונסולה ו-UI מוגבלות

השוו זאת לאפליקציה מבוססת פקודות בה מקלידים פקודה:

- **זה מוגבל**. אתם לא יכולים פשוט להקליד כל פקודה, אלא רק את אלו שהאפליקציה תומכת בהן.
- **מסוימות לשפה**. יש אפליקציות שתומכות בשפות רבות, אבל כברירת מחדל האפליקציה בנויה לשפה מסוימת, אפילו אם ניתן להוסיף תמיכה בשפות נוספות.

### יתרונות אפליקציות ליצירת טקסט

אז איך אפליקציית יצירת טקסט שונה?

באפליקציית יצירת טקסט יש לכם יותר גמישות, אתם לא מוגבלים למערך פקודות או לשפת קלט ספציפית. במקום זאת, אתם יכולים להשתמש בשפה טבעית כדי לתקשר עם האפליקציה. יתרון נוסף הוא שאתם כבר מתקשרים עם מקור נתונים שאומן על קורפוס עצום של מידע, בעוד שאפליקציה מסורתית יכולה להיות מוגבלת למה שמופיע בבסיס הנתונים.

### מה ניתן לבנות באפליקציית יצירת טקסט?

יש הרבה דברים שניתן לבנות. לדוגמה:

- **בוט צ'אט**. בוט צ'אט העונה על שאלות בנושאים כמו החברה שלכם והמוצרים שלה יכול להיות התאמה טובה.
- **עוזר**. LLMs טובים בדברים כמו סיכום טקסט, השגת תובנות מתוך טקסט, יצירת טקסט כמו קורות חיים ועוד.
- **עוזר קוד**. בהתאם למודל השפה שבו אתם משתמשים, ניתן לבנות עוזר קוד שמסייע לכם לכתוב קוד. לדוגמה, ניתן להשתמש במוצר כמו GitHub Copilot וגם ChatGPT כדי לסייע בכתיבת קוד.

### איתור מפתח ה-API והנקודת קצה

בשלב זה, עליכם לספר לספריית `openai` איזה מפתח API להשתמש בו. כדי למצוא את מפתח ה-API שלכם, עברו לסעיף "Keys and Endpoint" במשאב Azure OpenAI והעתיקו את הערך "Key 1".

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

כעת כשיש לכם את המידע הזה, ניתן להורות לספריות להשתמש בו.

> [!NOTE]
> כדאי להפריד את מפתח ה-API שלכם מהקוד. ניתן לעשות זאת באמצעות משתני סביבה.
> - הגדירו את משתנה הסביבה `AZURE_OPENAI_API_KEY` עם מפתח ה-API שלכם בקובץ .env. אם השלמתם את התרגילים הקודמים בקורס זה, אתם כבר מוגדרים.


### הגדרת תצורה לאז'ור

אם אתם משתמשים ב-Azure OpenAI, כך תגדירו את התצורה. ה-Responses API מופעל מנקודת הקצה **v1** של Azure OpenAI, לכן אנו מצביעים על לקוח ה-`OpenAI` ל-`<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

למעלה אנו מגדירים את הדברים הבאים:

- `api_key`, זה מפתח ה-API שלכם שנמצא בפורטל של Azure.
- `base_url`, זו נקודת הקצה שלכם ב-Azure OpenAI עם התוספת `/openai/v1/`. ניתן למצוא את נקודת הקצה בפורטל לצד מפתח ה-API. שימוש בנקודת הקצה v1 מפחית את הצורך להעביר `api_version`.
- `deployment`, זהו שם הפריסה של המודל שיצרתם בפורטל Foundry.

> [!NOTE]
> `os.environ` הוא מיפוי שנותן קריאה למשתני הסביבה. ניתן להשתמש בו כדי לקרוא משתני סביבה כמו `AZURE_OPENAI_API_KEY` ו-`AZURE_OPENAI_ENDPOINT`.

## יצירת טקסט

הדרך ליצירת טקסט היא באמצעות המתודה `responses.create`. הנה דוגמה:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

בקוד שלמעלה, אנו יוצרים אובייקט תגובה ומעבירים את המודל שאנחנו רוצים להשתמש בו ואת הפרומפט. לאחר מכן מדפיסים את הטקסט שנוצר.

### תגובות צ'אט

ה-Responses API מתאים גם ליצירת טקסט בפעם אחת וגם לבוטי צ'אט מרובי סיבובים - פשוט מספקים יותר הודעות ברשימת ה-`input` לבניית שיחה:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

מידע נוסף על פונקציונליות זו בפרק הבא.

## תרגיל - אפליקציית יצירת הטקסט הראשונה שלך

כעת שלמדנו כיצד להגדיר ולכוון את שירות Azure OpenAI, הגיע הזמן לבנות את אפליקציית יצירת הטקסט הראשונה שלך. כדי לבנות את האפליקציה, עקוב אחר השלבים הבאים:


1. צור סביבה וירטואלית והתקן את openai:

  > [!NOTE] שלב זה אינו נחוץ אם אתה מריץ פנקס זה ב-Codespaces או בתוך Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]

> [!NOTE]
> אתר את מפתח Azure OpenAI שלך על ידי כניסה ל- https://portal.azure.com/ וחיפוש אחרי `Open AI`, בחר את `Open AI resource` ואז בחר ב-`Keys and Endpoint` והעתק את הערך של `Key 1`.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## תרגיל - לבנות מחולל מתכונים

עכשיו כששיחקנו תרחיש, נכתוב קוד שיתאים לתרחיש שהודגם. לשם כך, פעלו לפי השלבים הבאים:

1. השתמשו בקובץ הקיים *app.py* כנקודת התחלה
1. מצאו את המשתנה `prompt` ושנו את הקוד שלו לקוד הבא:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?
