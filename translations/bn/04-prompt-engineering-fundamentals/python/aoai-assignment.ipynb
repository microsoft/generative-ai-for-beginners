# প্রম্পট ইঞ্জিনিয়ারিং পরিচিতি
প্রম্পট ইঞ্জিনিয়ারিং হল প্রকৃত ভাষা প্রক্রিয়াকরণ কাজের জন্য প্রম্পট ডিজাইন এবং অপ্টিমাইজ করার প্রক্রিয়া। এর মধ্যে সঠিক প্রম্পট নির্বাচন, তাদের প্যারামিটার টিউন করা এবং তাদের কার্যকারিতা মূল্যায়ন করা অন্তর্ভুক্ত। ন্যাচারাল ল্যাঙ্গুয়েজ প্রোসেসিং (NLP) মডেলগুলিতে উচ্চ সঠিকতা এবং দক্ষতা অর্জনের জন্য প্রম্পট ইঞ্জিনিয়ারিং অপরিহার্য। এই বিভাগে, আমরা OpenAI মডেলগুলি ব্যবহার করে প্রম্পট ইঞ্জিনিয়ারিংয়ের বুনিয়াদি অন্বেষণ করব।


### অনুশীলন ১: টোকেনাইজেশন  
ওপেনএআই থেকে একটি ওপেন-সোর্স দ্রুত টোকেনাইজার tiktoken ব্যবহার করে টোকেনাইজেশন অন্বেষণ করুন  
আরও উদাহরণের জন্য দেখুন [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) ।  


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### ব্যায়াম ২: OpenAI API কী সেটআপ যাচাই

আপনার OpenAI এন্ডপয়েন্ট সঠিকভাবে সেটআপ হয়েছে কি না তা যাচাই করার জন্য নিচের কোডটি চালান। কোডটি একটি সাধারণ বেসিক প্রম্পট দেয় এবং কমপ্লিশন যাচাই করে। ইনপুট `oh say can you see` এর জন্য আউটপুট হওয়া উচিত `by the dawn's early light..` এর মতো। 


````python
# Uses the OpenAI client against the Azure OpenAI (Microsoft Foundry) v1 endpoint
# with the Responses API. See https://aka.ms/openai/start

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### অনুশীলন ৩: গঠনমূলক তথ্য
যখন আপনি LLM-কে এমন একটি বিষয়ের জন্য প্রম্পট অনুযায়ী সম্পূর্ণতা ফেরত দিতে বলেন যা থাকতে পারে না, অথবা এমন বিষয়গুলি সম্পর্কে যা এটি জানতে নাও পারে কারণ তা এর প্রাক-প্রশিক্ষিত ডেটাসেটের বাইরে ছিল (সাম্প্রতিক), তখন কী ঘটে তা পরীক্ষা করুন। দেখুন উত্তরের পরিবর্তন কেমন হয় যদি আপনি একটি ভিন্ন প্রম্পট বা একটি ভিন্ন মডেল ব্যবহার করেন।


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### অনুশীলন ৪: নির্দেশনা ভিত্তিক  
"text" ভেরিয়েবলটি প্রাথমিক বিষয়বস্তু সেট করতে ব্যবহার করুন  
এবং "prompt" ভেরিয়েবলটি ঐ প্রাথমিক বিষয়বস্তু সম্পর্কিত একটি নির্দেশনা দিতে ব্যবহার করুন।  

এখানে আমরা মডেলকে অনুরোধ করছি দ্বিতীয় শ্রেণির ছাত্রছাত্রীদের জন্য পাঠ্যটি সারাংশ করতে  


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### অনুশীলন ৫: জটিল প্রম্পট  
একটি অনুরোধ চেষ্টা করুন যেখানে সিস্টেম, ব্যবহারকারী এবং সহকারী বার্তাগুলি থাকবে  
সিস্টেম সহকারীর প্রসঙ্গ নির্ধারণ করে  
ব্যবহারকারী ও সহকারী বার্তাগুলি বহু-পর্বের আলাপচারিতার প্রসঙ্গ প্রদান করে  

লক্ষ্য করুন কিভাবে সহকারীর ব্যক্তিত্ব "বিদ্রূপাত্মক" হিসেবে সিস্টেম প্রসঙ্গে সেট করা হয়েছে।  
ভিন্ন ব্যক্তিত্ব প্রসঙ্গ ব্যবহার করে দেখুন। অথবা ইনপুট/আউটপুট বার্তাগুলোর ভিন্ন সিরিজ চেষ্টা করুন  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
