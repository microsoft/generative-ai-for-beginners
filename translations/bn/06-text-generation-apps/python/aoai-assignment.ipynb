# পাঠ্য উৎপাদন অ্যাপ তৈরি করুন

এই পাঠক্রমে আপনি এখন পর্যন্ত দেখেছেন যে প্রম্পটের মতো মূল ধারণাগুলো রয়েছে এবং এমনকি "প্রম্পট ইঞ্জিনিয়ারিং" নামে একটি সম্পূর্ণ শাখাও আছে। অনেক সরঞ্জাম যেমন ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক কিছু আপনাকে কিছু অর্জন করতে প্রম্পট ব্যবহার করার মাধ্যমে সাহায্য করে।

এমন একটি অভিজ্ঞতা একটি অ্যাপে যোগ করতে, আপনাকে প্রম্পট, সমাপ্তি ইত্যাদি ধারণাগুলো বোঝা এবং ব্যবহারের জন্য একটি লাইব্রেরি বেছে নেওয়া প্রয়োজন। এটাই আপনি এই অধ্যায়ে শিখবেন।

## পরিচিতি

এই অধ্যায়ে আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলো সম্পর্কে শিখবেন।
- openai ব্যবহার করে একটি পাঠ্য উৎপাদন অ্যাপ তৈরি করবেন।
- প্রম্পট, তাপমাত্রা, এবং টোকেনের মতো ধারণাগুলি কীভাবে ব্যবহার করা যায় তা বুঝবেন।

## শেখার লক্ষ্য

এই পাঠ শেষ হলে, আপনি সক্ষম হবেন:

- একটি পাঠ্য উৎপাদন অ্যাপ কি তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি পাঠ্য উৎপাদন অ্যাপ তৈরি করতে।
- আপনার অ্যাপকে কম বা বেশী টোকেন ব্যবহারের জন্য কনফিগার করতে এবং আউটপুট বৈচিত্র্যের জন্য তাপমাত্রাও পরিবর্তন করতে।

## পাঠ্য উৎপাদন অ্যাপ কি?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, তার একটি ইন্টারফেস থাকে যেমন নিম্নলিখিত:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপগুলি হলো এমন অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং তা একটি কাজ সম্পন্ন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ব্যবহারকারী ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিক্যাল ব্যবহারকারী ইন্টারফেস (GUI) থাকে যেখানে আপনি বাটন ক্লিক করেন, টেক্সট ইনপুট করেন, অপশন নির্বাচন করেন এবং আরও অনেক কিছু।

### কনসোল এবং UI অ্যাপের সীমাবদ্ধতা

একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমাবদ্ধ**। আপনি যে কোনও কমান্ড টাইপ করতে পারবেন না, শুধুমাত্র সেই গুলো যা অ্যাপ সমর্থন করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সমর্থন করে, তবে ডিফল্টরূপে অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য নির্মিত, যদিও আপনি আরও ভাষা যোগ করতে পারেন।

### পাঠ্য উৎপাদন অ্যাপের সুবিধা

তাহলে পাঠ্য উৎপাদন অ্যাপটি কীভাবে আলাদা?

একটি পাঠ্য উৎপাদন অ্যাপে, আপনার বেশি নমনীয়তা থাকে, আপনি নির্দিষ্ট কিছু কমান্ড বা নির্দিষ্ট ইনপুট ভাষার মধ্যে আবদ্ধ থাকেন না। বরং, আপনি প্রাকৃতিক ভাষা ব্যবহার করে অ্যাপটির সাথে ইন্টারঅ্যাক্ট করতে পারেন। অন্য একটি সুবিধা হল আপনি ইতিমধ্যে এমন একটি তথ্য উৎসের সাথে ইন্টারঅ্যাক্ট করছেন যা একটি বিশাল তথ্য সংগৃহীত কপাসে প্রশিক্ষিত, যেখানে ঐতিহ্যবাহী অ্যাপগুলি সাধারণত ডাটাবেজে সীমাবদ্ধ থাকে।

### পাঠ্য উৎপাদন অ্যাপ দিয়ে আমি কী তৈরি করতে পারি?

আপনি অনেক কিছু তৈরি করতে পারেন। উদাহরণস্বরূপ:

- **একটি চ্যাটবট**। একটি চ্যাটবট যা বিভিন্ন বিষয়ে প্রশ্নের উত্তর দেয়, যেমন আপনার কোম্পানি এবং এর পণ্য সম্পর্কে, এটি একটি ভাল মিল হতে পারে।
- **সহায়ক**। LLM গুলো লেখার সারাংশ তৈরি করা, লেখার থেকে তথ্য সংগ্রহ করা, রেজুমি তৈরি করা এবং আরও অনেক ক্ষেত্রে দুর্দান্ত হতে পারে।
- **কোড সহকারী**। আপনি যে ভাষা মডেল ব্যবহার করেন তার ওপর নির্ভর করে, আপনি এমন একটি কোড সহকারী তৈরি করতে পারেন যা কোড লেখায় সাহায্য করে। উদাহরণস্বরূপ, GitHub Copilot এবং ChatGPT এর মতো পণ্য ব্যবহার করে কোড লেখায় সাহায্য নিতে পারেন।

## আমি কীভাবে শুরু করতে পারি?

সাধারণত, LLM এর সাথে সংযুক্ত হতে নিম্নলিখিত দুটি পদ্ধতি রয়েছে:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পট সহ ওয়েব রিকোয়েস্ট তৈরি করেন এবং উৎপাদিত পাঠ্য ফিরে পান।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলি API কলগুলোকে এনক্যাপসুলেট করে এবং ব্যবহার সহজ করে তোলে।

## লাইব্রেরি/SDK

LLM এর সাথে কাজ করার জন্য কিছু সুপরিচিত লাইব্রেরি আছে যেমন:

- **openai**, এই লাইব্রেরি আপনার মডেলের সাথে সংযোগ স্থাপন করা এবং প্রম্পট পাঠানো সহজ করে।

এরপর কিছু লাইব্রেরি আছে যা উচ্চতর স্তরে কাজ করে যেমন:

- **Langchain**। Langchain সুপরিচিত একটি লাইব্রেরি এবং এটি Python সমর্থন করে।
- **Semantic Kernel**। Semantic Kernel Microsoft এর একটি লাইব্রেরি যা C#, Python, এবং Java ভাষাগুলো সমর্থন করে।

## openai ব্যবহার করে প্রথম অ্যাপ

চলুন দেখি আমরা কীভাবে প্রথম অ্যাপ তৈরি করতে পারি, কী লাইব্রেরি দরকার, কত কিছু প্রয়োজন ইত্যাদি।

### openai ইনস্টল করুন

  > [!NOTE] যদি আপনি Codespaces বা Devcontainer এ এই নোটবুক চালান, এই ধাপ প্রয়োজন নেই


OpenAI বা Azure OpenAI এর সাথে কাজ করার জন্য অসংখ্য লাইব্রেরি রয়েছে। C#, Python, JavaScript, Java এবং আরও অনেক ভাষা ব্যবহার করেও কাজ করা যায়।  
আমরা `openai` Python লাইব্রেরিটি ব্যবহার করার সিদ্ধান্ত নিয়েছি, তাই আমরা `pip` ব্যবহার করে এটি ইনস্টল করব।

```bash
pip install openai
```

যদি আপনি Codespaces অথবা Dev Container এ এই নোটবুক না চালান, তাহলে আপনাকে আপনার মেশিনে [Python](https://www.python.org/) ইনস্টলও করতে হবে।

### একটি রিসোর্স তৈরি করুন

যদি আপনি আগে না করে থাকেন, তাহলে নিম্নলিখিত পদক্ষেপগুলো সম্পন্ন করতে হবে:

- Azure এ একটি অ্যাকাউন্ট তৈরি করুন <https://azure.microsoft.com/free/>
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) তে একটি Azure OpenAI রিসোর্স তৈরি করুন। কিভাবে [রিসোর্স তৈরি করবেন তা দেখতে](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) এই গাইডটি দেখুন।


### API কী এবং এন্ডপয়েন্ট লোকেট করুন

এই পর্যায়ে, আপনাকে আপনার `openai` লাইব্রেরিকে কী API কী ব্যবহার করতে হবে তা বলতে হবে। আপনার API কী খুঁজে পেতে Azure OpenAI রিসোর্সের "Keys and Endpoint" বিভাগে যান এবং "Key 1" এর মান কপি করুন।

  ![Azure Portal এ Keys এবং Endpoint রিসোর্স ব্লেড](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

এখন এই তথ্য কপি করা হয়েছে, চলুন লাইব্রেরিগুলোকে এটি ব্যবহার করার নির্দেশনা দিই।

> [!NOTE]
> আপনার API কী কোড থেকে আলাদা রাখা ভাল। আপনি পরিবেশ ভেরিয়েবল ব্যবহার করে এটি করতে পারেন।
> - `AZURE_OPENAI_API_KEY` পরিবেশ ভেরিয়েবল আপনার API কী দিয়ে .env ফাইলে সেট করুন। আপনি যদি এই কোর্সের পূর্ববর্তী অনুশীলনগুলি করেছেন, তাহলে এটি ইতিমধ্যে সেট আপ করা আছে।


### Azure কনফিগারেশন সেট আপ করুন

যদি আপনি Azure OpenAI ব্যবহার করেন, তাহলে কনফিগারেশন সেট আপ এর পদ্ধতি হলো: Responses API Azure OpenAI **v1 endpoint** থেকে সেবা দেয়, সুতরাং `OpenAI` ক্লায়েন্টকে `<your-endpoint>/openai/v1/` এ নির্দেশ দেয়া হয়:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

উপরে আমরা নিম্নলিখিতগুলো সেট করছি:

- `api_key`, এটি আপনার Azure Portal এ পাওয়া API কী।
- `base_url`, এটি আপনার Azure OpenAI এন্ডপয়েন্ট `/openai/v1/` যুক্ত করে। এন্ডপয়েন্টটি আপনার API কী পাশে Azure Portal এ পাওয়া যাবে। v1 এন্ডপয়েন্ট ব্যবহার মানে এখন আর `api_version` দিতে হবে না।
- `deployment`, এটি Foundry পোর্টালে আপনি তৈরি করা মডেল ডিপ্লয়মেন্টের নাম।

> [!NOTE]
> `os.environ` হলো একটি ম্যাপ যা পরিবেশ ভেরিয়েবল পড়ে। আপনি এটি ব্যবহার করে `AZURE_OPENAI_API_KEY` এবং `AZURE_OPENAI_ENDPOINT` এর মত পরিবেশ ভেরিয়েবল পড়তে পারেন।

## পাঠ্য তৈরি করুন

পাঠ্য তৈরি করার পদ্ধতি হলো `responses.create` মেথড ব্যবহার করা। উদাহরণ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

উপরের কোডে, আমরা একটি response অবজেক্ট তৈরি করি এবং আমরা যে মডেল ব্যবহার করতে চাই এবং প্রম্পট সেট করি। এরপর উৎপাদিত পাঠ্য প্রিন্ট করি।

### চ্যাট প্রতিক্রিয়া

Responses API একক-টর পাঠ্য উৎপাদন এবং বহু-টর চ্যাটবটের জন্য উপযুক্ত - আপনি কেবল `input` তালিকায় আরও বার্তা প্রদান করে একটি কথোপকথন তৈরি করতে পারেন:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

এই কার্যকারিতা সম্পর্কে আরও একটি অধ্যায়ে আলোচনা হবে।

### Improve by adding filter and shopping list

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        ```
