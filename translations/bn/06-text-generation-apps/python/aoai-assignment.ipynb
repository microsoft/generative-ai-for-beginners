# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

আপনি এখন পর্যন্ত এই পাঠক্রমে দেখেছেন যে প্রম্পটের মতো মুল ধারণাগুলো আছে এবং এমনকি একটি সম্পূর্ণ শৃঙ্খলা আছে যাকে বলা হয় "প্রম্পট ইঞ্জিনিয়ারিং"। আপনি অনেক সরঞ্জামের সাথে ইন্টারঅ্যাক্ট করতে পারেন যেমন ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক কিছু, যা আপনাকে প্রম্পট ব্যবহার করে কিছু অর্জন করার জন্য সমর্থন করে।

আপনাকে যদি এমন একটি অভিজ্ঞতা একটি অ্যাপে যোগ করতে হয়, তাহলে আপনাকে প্রম্পট, কমপ্লিশন সহ ধারণাগুলো বুঝতে হবে এবং কাজ করার জন্য একটি লাইব্রেরি নির্বাচন করতে হবে। ঠিক এইটাই আপনি এই অধ্যায়ে শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলি সম্পর্কে শিখবেন।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, তাপমাত্রা, এবং টোকেনের মতো ধারণাগুলো ব্যবহার করে কীভাবে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন তা বুঝবেন।

## শেখার লক্ষ্যসমূহ

এই পাঠের শেষে, আপনি সক্ষম হবেন:

- টেক্সট জেনারেশন অ্যাপ কি তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপ কনফিগার করতে যাতে কম বা বেশি টোকেন ব্যবহার হয় এবং তাপমাত্রা পরিবর্তন করা যায়, বিভিন্ন ধরনের আউটপুটের জন্য।

## টেক্সট জেনারেশন অ্যাপ কি?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন তখন এর কিছু ধরনের ইন্টারফেস থাকে যেমন নিম্নলিখিত:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপগুলি সাধারণত কমান্ড টাইপ করে একটি কাজ সম্পাদন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ব্যবহারকারী ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিক্যাল ইউজার ইন্টারফেস (GUI) থাকে যেখানে আপনি বোতাম ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন নির্বাচন করেন এবং আরও অনেক কিছু করতে পারেন।

### কনসোল এবং UI অ্যাপগুলি সীমাবদ্ধ

এর সাথে তুলনা করুন একটি কমান্ড-ভিত্তিক অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমাবদ্ধ**। আপনি যেকোনো কমান্ড টাইপ করতে পারবেন না, শুধু সেসব কমান্ডই যা অ্যাপ সমর্থন করে।
- **ভাষাভিত্তিক**। কিছু অ্যাপ অনেক ভাষা সমর্থন করে, তবে ডিফল্টরূপে অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি আরও ভাষা সংযোজন করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধাসমূহ

তাহলে টেক্সট জেনারেশন অ্যাপের পার্থক্য কী?

একটি টেক্সট জেনারেশন অ্যাপ আপনাকে আরও নমনীয়তা দেয়, আপনি একটি নির্দিষ্ট কমান্ড বা ইনপুট ভাষার মধ্যে সীমাবদ্ধ থাকেন না। পরিবর্তে, আপনি অ্যাপের সাথে প্রাকৃতিক ভাষায় ইন্টারঅ্যাক্ট করতে পারেন। আরও একটি সুবিধা হলো যে আপনি ইতিমধ্যে এমন একটি তথ্য উৎসের সাথে সংযুক্ত আছেন যা বিশাল তথ্যভাণ্ডারে প্রশিক্ষিত, যেখানে একটি ঐতিহ্যবাহী অ্যাপ ডাটাবেজের মধ্যে সীমাবদ্ধ থাকতে পারে।

### টেক্সট জেনারেশন অ্যাপ দিয়ে আমি কী তৈরি করতে পারি?

আপনি অনেক কিছু তৈরি করতে পারেন। যেমন:

- **একটি চ্যাটবট**। একটি চ্যাটবট যা আপনার কোম্পানি এবং এর পণ্য সম্পর্কে প্রশ্নের উত্তর দেয় ভাল একটি মিল হতে পারে।
- **সহায়ক**। LLM গুলো টেক্সট সংক্ষিপ্তকরণ, টেক্সট থেকে তথ্য সংগ্রহ, রিজুমে মত টেক্সট তৈরি ইত্যাদিতে দুর্দান্ত।
- **কোড সহকারী**। আপনি যে ভাষার মডেল ব্যবহার করছেন তার ওপর নির্ভর করে, আপনি এমন এক কোড সহকারী তৈরি করতে পারেন যা আপনাকে কোড লিখতে সাহায্য করে। উদাহরণস্বরূপ, আপনি GitHub Copilot এর মতো প্রোডাক্ট বা ChatGPT ব্যবহার করে কোড লেখায় সাহায্য পেতে পারেন।

## আমি কীভাবে শুরু করব?

ভাল, আপনাকে LLM এর সাথে ইন্টিগ্রেট করার একটি উপায় খুঁজে বের করতে হবে যা সাধারণত নিম্নলিখিত দুইটি পদ্ধতি অন্তর্ভুক্ত করে:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পট দিয়ে ওয়েব রিকোয়েস্ট তৈরি করেন এবং জেনারেটেড টেক্সট পেয়ে থাকেন।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলো API কলগুলিকে সংজ্ঞায়িত করে এবং ব্যবহারে সহজ করে তোলে।

## লাইব্রেরি/SDK

LLM এর সাথে কাজ করার জন্য কয়েকটি পরিচিত লাইব্রেরি রয়েছে যেমন:

- **openai**, এই লাইব্রেরি আপনার মডেলের সাথে সহজে সংযোগ করতে এবং প্রম্পট পাঠাতে সাহায্য করে।

তারপর আরো উপরের স্তরের লাইব্রেরি রয়েছে যেমন:

- **Langchain**। Langchain পরিচিত এবং Python সমর্থন করে।
- **Semantic Kernel**। Semantic Kernel মাইক্রোসফটের একটি লাইব্রেরি যা C#, Python, এবং Java ভাষাগুলো সমর্থন করে।

## openai দিয়ে প্রথম অ্যাপ

চলুন দেখি কীভাবে আমরা আমাদের প্রথম অ্যাপ তৈরি করতে পারি, কোন লাইব্রেরি দরকার এবং কতটা প্রয়োজন তাদের সম্পর্কে।

### openai ইনস্টল করুন

  > [!NOTE] Codespaces বা Devcontainer এ এই নোটবুক চালালে এই ধাপটি জরুরি নয়


OpenAI বা Azure OpenAI এর সাথে ইন্টারঅ্যাক্ট করার জন্য অনেক লাইব্রেরি আছে। C#, Python, JavaScript, Java সহ বিভিন্ন প্রোগ্রামিং ভাষাও ব্যবহার করা যায়।  
আমরা `openai` Python লাইব্রেরি ব্যবহার করার সিদ্ধান্ত নিয়েছি, তাই আমরা `pip` দিয়ে এটি ইনস্টল করব।

```bash
pip install openai
```

যদি আপনি এই নোটবুক Codespaces বা Dev Container এ না চালান, তাহলে আপনাকে আপনার মেশিনে [Python](https://www.python.org/) ইনস্টল করতে হবে।

### একটি রিসোর্স তৈরি করুন

যদি আপনি আগে তৈরি না করে থাকেন, আপনাকে নিম্নলিখিত ধাপগুলো অনুসরণ করতে হবে:

- <https://azure.microsoft.com/free/> এ একটি Azure অ্যাকাউন্ট তৈরি করুন।
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) এ একটি Azure OpenAI রিসোর্স তৈরি করুন। কীভাবে [রিসোর্স তৈরি করবেন](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) তার জন্য এই গাইড দেখুন।


### API কী এবং এন্ডপয়েন্ট সন্ধান করুন

এই পর্যায়ে আপনাকে আপনার `openai` লাইব্রেরিকে কোন API কী ব্যবহার করতে হবে তা জানাতে হবে। আপনার API কী জানতে, আপনার Azure OpenAI রিসোর্স এর "Keys and Endpoint" বিভাগে যান এবং "Key 1" মান কপি করুন।

  ![Azure পোর্টালে Keys and Endpoint রিসোর্স ব্লেড](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

এখন আপনি এই তথ্যটি কপি করে নিয়েছেন, আসুন লাইব্রেরিগুলোকে এটি ব্যবহার করতে নির্দেশ দিই।

> [!NOTE]
> আপনার API কী কোড থেকে আলাদা রাখা ভাল। আপনি এটি পরিবেশ ভেরিয়েবল ব্যবহার করে করতে পারেন।
> - পরিবেশ ভেরিয়েবল `AZURE_OPENAI_API_KEY` আপনার .env ফাইলে আপনার API কী হিসাবে সেট করুন। যদি আপনি ইতিমধ্যে এই কোর্সের পূর্ববর্তী অনুশীলন সম্পন্ন করে থাকেন, তবে আপনি প্রস্তুত।


### Azure কনফিগারেশন সেটআপ

আপনি যদি Azure OpenAI ব্যবহার করেন, তাহলে কনফিগারেশন এইরকম করবেন। Responses API Azure OpenAI **v1 endpoint** থেকে পরিবেশিত হয়, তাই আমরা `OpenAI` ক্লায়েন্টকে `<your-endpoint>/openai/v1/` তে পয়েন্ট করি:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

উপরে আমরা নিম্নলিখিতগুলি সেট করছি:

- `api_key`, এটি আপনার API কী যা Azure পোর্টালে পাওয়া গেছে।
- `base_url`, এটি আপনার Azure OpenAI এন্ডপয়েন্ট `/openai/v1/` যুক্ত করে। আপনি Azure পোর্টালে আপনার API কী-এর পাশে এন্ডপয়েন্টটি দেখতে পাবেন। v1 এন্ডপয়েন্ট ব্যবহার মানে আপনাকে আর `api_version` পাঠাতে হবে না।
- `deployment`, এটি Foundry পোর্টালে আপনি যে মডেল ডিপ্লয়মেন্ট তৈরি করেছেন তার নাম।

> [!NOTE]
> `os.environ` একটি ম্যাপ যা পরিবেশ ভেরিয়েবলগুলি পড়ে। আপনি এটি ব্যবহার করে `AZURE_OPENAI_API_KEY` এবং `AZURE_OPENAI_ENDPOINT` এর মতো পরিবেশ ভেরিয়েবল পড়তে পারেন।

## টেক্সট তৈরি করুন

টেক্সট তৈরি করার পদ্ধতি হলো `responses.create` মেথড ব্যবহার করা। উদাহরণ আছে এখানে:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

উপরের কোডে, আমরা একটি রেসপন্স অবজেক্ট তৈরি করি এবং আমরা যে মডেল ব্যবহার করতে চাই এবং প্রম্পট পাস করি। তারপর আমরা জেনারেট করা টেক্সট প্রিন্ট করি।

### চ্যাট রেসপন্স

Responses API একক-ট্যুর টেক্সট জেনারেশন এবং বহু-ট্যুর চ্যাটবটের জন্য ভালোভাবে উপযুক্ত - আপনি যেকোন সময়ের কথোপকথনের জন্য `input` তালিকায় আরও বার্তা দিতে পারেন:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
```
