# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

এই কারিকুলামে এখন পর্যন্ত আপনি দেখেছেন যে প্রম্পটের মতো কিছু মূল ধারণা আছে এবং এমনকি "প্রম্পট ইঞ্জিনিয়ারিং" নামে একটি সম্পূর্ণ শাখাও আছে। আপনি যেসব টুলের সাথে কাজ করতে পারেন যেমন ChatGPT, Office 365, Microsoft Power Platform ইত্যাদি, সেগুলো আপনাকে কিছু করার জন্য প্রম্পট ব্যবহার করতে সহায়তা করে।

আপনি যদি আপনার অ্যাপে এমন অভিজ্ঞতা যোগ করতে চান, তাহলে আপনাকে প্রম্পট, কমপ্লিশন এবং কোন লাইব্রেরি ব্যবহার করবেন তা জানতে হবে। এই অধ্যায়ে আপনি ঠিক এগুলোই শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি শিখবেন:

- openai লাইব্রেরি এবং এর মূল ধারণা সম্পর্কে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, টেম্পারেচার এবং টোকেনের মতো ধারণা ব্যবহার করে কিভাবে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন তা বুঝবেন।

## শেখার লক্ষ্য

এই পাঠ শেষে, আপনি পারবেন:

- টেক্সট জেনারেশন অ্যাপ কী তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপে কত টোকেন ব্যবহার হবে এবং টেম্পারেচার পরিবর্তন করে আউটপুটে বৈচিত্র্য আনতে কনফিগার করতে।

## টেক্সট জেনারেশন অ্যাপ কী?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, তার কিছু ইন্টারফেস থাকে যেমন:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপ সাধারণত এমন অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং এটি একটি কাজ করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ইউজার ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিক্যাল ইউজার ইন্টারফেস (GUI) থাকে যেখানে আপনি বাটনে ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন সিলেক্ট করেন ইত্যাদি।

### কনসোল ও UI অ্যাপের সীমাবদ্ধতা

এটি একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমিত**। আপনি যেকোনো কমান্ড টাইপ করতে পারবেন না, শুধু যেগুলো অ্যাপটি সাপোর্ট করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সাপোর্ট করে, কিন্তু সাধারণত অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি আরও ভাষা যোগ করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে টেক্সট জেনারেশন অ্যাপ কীভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে আপনার আরও স্বাধীনতা থাকে, আপনি নির্দিষ্ট কিছু কমান্ড বা ইনপুট ভাষার মধ্যে সীমাবদ্ধ নন। বরং, আপনি স্বাভাবিক ভাষা ব্যবহার করে অ্যাপের সাথে কথা বলতে পারেন। আরেকটি সুবিধা হলো, আপনি এমন একটি ডেটা সোর্সের সাথে কাজ করছেন যা বিশাল তথ্যভাণ্ডারে প্রশিক্ষিত, যেখানে সাধারণ অ্যাপ কেবল ডাটাবেসের তথ্যেই সীমাবদ্ধ।

### টেক্সট জেনারেশন অ্যাপ দিয়ে কী তৈরি করা যায়?

অনেক কিছু তৈরি করা যায়। যেমন:

- **চ্যাটবট**। চ্যাটবট যেটি আপনার কোম্পানি ও পণ্যের মতো বিষয় নিয়ে প্রশ্নের উত্তর দিতে পারে।
- **সহায়ক**। LLM গুলো টেক্সট সারাংশ করা, টেক্সট থেকে ইনসাইট বের করা, রিজিউমের মতো টেক্সট তৈরি করা ইত্যাদিতে দারুণ।
- **কোড সহকারী**। আপনি যে ভাষার মডেল ব্যবহার করছেন তার উপর ভিত্তি করে, আপনি এমন কোড সহকারী তৈরি করতে পারেন যা কোড লিখতে সাহায্য করে। যেমন, GitHub Copilot বা ChatGPT ব্যবহার করে কোড লেখার সহায়তা নিতে পারেন।

## কিভাবে শুরু করব?

আপনাকে LLM-এর সাথে ইন্টিগ্রেট করার উপায় খুঁজতে হবে, সাধারণত দুটি উপায়ে:

- API ব্যবহার করুন। এখানে আপনি প্রম্পট সহ ওয়েব রিকোয়েস্ট তৈরি করেন এবং জেনারেটেড টেক্সট ফিরে পান।
- লাইব্রেরি ব্যবহার করুন। লাইব্রেরি API কলগুলো সহজ করে তোলে এবং ব্যবহার সহজ হয়।

## লাইব্রেরি/SDK

LLM নিয়ে কাজ করার জন্য কিছু পরিচিত লাইব্রেরি আছে যেমন:

- **openai**, এই লাইব্রেরি দিয়ে সহজেই আপনার মডেলের সাথে কানেক্ট করে প্রম্পট পাঠানো যায়।

এছাড়াও কিছু উচ্চ-স্তরের লাইব্রেরি আছে যেমন:

- **Langchain**। Langchain বেশ পরিচিত এবং Python সাপোর্ট করে।
- **Semantic Kernel**। Semantic Kernel Microsoft-এর একটি লাইব্রেরি, C#, Python, এবং Java সাপোর্ট করে।

### ১. Codespaces 

* 'Get started' উইন্ডো থেকে "Run codespace" বেছে নিন
* একটি নতুন codespace তৈরি করুন (বা আগেরটি ব্যবহার করুন)
* VS Code আপনার ব্রাউজারে খুলবে, যেখানে বিভিন্ন ভাষার নমুনা নোটবুক পাবেন

> নোট: Codespaces-এ Github Token ভ্যারিয়েবল সেট করার দরকার নেই, এই ধাপটি বাদ দিন

**এখন নিচের 'Generate Text' সেকশনে যান এই অ্যাসাইনমেন্ট চালিয়ে যেতে**

### ২. VS Code (বা আপনার পছন্দের IDE)

'Get started' সবুজ বাটনে ক্লিক করলে আপনার পছন্দের IDE-তে চালানোর জন্য সব তথ্য পাবেন। এখানে VS Code দেখানো হয়েছে

* ভাষা ও SDK নির্বাচন করুন - এখানে Python এবং Azure AI Inference SDK বেছে নেওয়া হয়েছে
* GitHub-এ একটি personal access token তৈরি করুন। এটি Developer Settings-এ থাকে। টোকেনের জন্য কোনো পারমিশন দিতে হবে না। মনে রাখবেন, টোকেনটি Microsoft সার্ভিসে পাঠানো হবে।
* আপনার Github personal access token সংরক্ষণের জন্য একটি environment variable তৈরি করুন - bash, powershell এবং windows command prompt-এ নমুনা দেওয়া আছে
* ডিপেন্ডেন্সি ইনস্টল করুন: ```pip install azure-ai-inference```
* মূল নমুনা কোডটি একটি .py ফাইলে কপি করুন
* যেখানে কোডটি সংরক্ষণ করেছেন সেখানে যান এবং ফাইলটি চালান: ```python filename.py```

মনে রাখবেন, Azure AI Inference SDK ব্যবহার করে আপনি সহজেই কোডে `model_name` পরিবর্তন করে বিভিন্ন মডেল নিয়ে পরীক্ষা করতে পারেন।

- **নির্দিষ্ট ধরণের টেক্সট তৈরি করা**। যেমন, আপনি একটি কবিতা, কুইজের জন্য প্রশ্ন ইত্যাদি তৈরি করতে পারেন।

- **কোড তৈরি করা**। আপনি প্রম্পট ব্যবহার করে কোড তৈরি করতে পারেন, যেমন ইমেইল যাচাই করার জন্য একটি রেগুলার এক্সপ্রেশন তৈরি করা, কিংবা পুরো একটি প্রোগ্রাম, যেমন একটি ওয়েব অ্যাপ তৈরি করা।

## অনুশীলন - একটি রেসিপি জেনারেটর তৈরি করুন

এখন যেহেতু আমরা একটি পরিস্থিতি দেখে নিয়েছি, চলুন সেই পরিস্থিতির সাথে মিল রেখে কোড লিখি। এটি করতে, নিচের ধাপগুলো অনুসরণ করুন:

1. বিদ্যমান ফাইলটি শুরু হিসেবে ব্যবহার করুন
1. একটি `prompt` ভেরিয়েবল তৈরি করুন এবং নমুনা কোডটি নিচের মতো পরিবর্তন করুন:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

এখন যদি আপনি কোডটি চালান, তাহলে আপনি নিচের মতো একটি আউটপুট দেখতে পাবেন:

দারুণ, চলুন দেখি কীভাবে আরও ভালো করা যায়। আরও ভালো করতে চাইলে, আমাদের নিশ্চিত করতে হবে কোডটি যেন নমনীয় হয়, যাতে উপাদান এবং রেসিপির সংখ্যা সহজেই পরিবর্তন ও উন্নত করা যায়।


চলুন নিম্নলিখিতভাবে কোডটি পরিবর্তন করি:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

    এই ফিচারটির জন্য, আমরা চাইলে সবকিছু এক প্রম্পটে করতে পারি অথবা দুইটি প্রম্পটে ভাগ করে নিতে পারি। চলুন দ্বিতীয় পদ্ধতিটি চেষ্টা করি। এখানে আমরা একটি অতিরিক্ত প্রম্পট যোগ করার কথা বলছি, তবে এটি কাজ করতে হলে, প্রথম প্রম্পটের ফলাফলটি দ্বিতীয় প্রম্পটের কনটেক্সট হিসেবে যোগ করতে হবে।

    কোডের সেই অংশটি খুঁজে বের করুন যেখানে প্রথম প্রম্পটের ফলাফল প্রিন্ট করা হয় এবং নিচের কোডটি সেখানে যোগ করুন:
    

    - আমরা একটি নতুন অনুরোধ করছি, তবে প্রথম প্রম্পটে যত টোকেন চেয়েছিলাম সেটিও বিবেচনা করছি, তাই এবার `max_tokens` ১২০০ বলেছি। **টোকেন দৈর্ঘ্য নিয়ে কিছু কথা**। আমাদের কতগুলো টোকেন প্রয়োজন তা বিবেচনা করা উচিত, কারণ টোকেনের জন্য খরচ হয়, তাই যেখানে সম্ভব, কম টোকেন ব্যবহার করার চেষ্টা করা উচিত। উদাহরণস্বরূপ, আমরা কি প্রম্পটটি এমনভাবে সাজাতে পারি যাতে কম টোকেন লাগে?
