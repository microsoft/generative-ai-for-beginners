# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

আপনি এতদূর এই পাঠ্যক্রমে দেখেছেন যে প্রম্পটের মতো মূল ধারণা রয়েছে এবং এমনকি "প্রম্পট ইঞ্জিনিয়ারিং" নামে একটি সম্পূর্ণ শাখা রয়েছে। আপনি অনেক সরঞ্জামের সাথে ইন্টারঅ্যাক্ট করতে পারেন যেমন ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক, যা আপনাকে কিছু সম্পাদনের জন্য প্রম্পট ব্যবহার করার সুবিধা দেয়।

আপনাকে যদি এমন একটি অভিজ্ঞতা একটি অ্যাপে যুক্ত করতে হয়, তবে আপনাকে প্রম্পট, কমপ্লিশনসের মতো ধারণাগুলি বুঝতে হবে এবং কাজ করার জন্য একটি লাইব্রেরি বেছে নিতে হবে। ঠিক یہی আপনি এই অধ্যায়ে শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলি সম্পর্কে জানবেন।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, তাপমাত্রা, এবং টোকেনের মতো ধারণাগুলি কীভাবে ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করা যায় তা বুঝবেন।

## শেখার লক্ষ্য

এই পাঠের শেষ পর্যন্ত, আপনি সক্ষম হবেন:

- টেক্সট জেনারেশন অ্যাপ কী তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপকে বেশি বা কম টোকেন ব্যবহার করতে কনফিগার করতে এবং তাপমাত্রা পরিবর্তন করতে সক্ষম হবেন, যাতে ভিন্ন আউটপুট পাওয়া যায়।

## টেক্সট জেনারেশন অ্যাপ কী?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, তখন এটি কিছু ধরনের ইন্টারফেস থাকে যেমন নিচের মতো:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপ গুলো সাধারণত এমন অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং এটি একটি কাজ সম্পন্ন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ব্যবহারকারী ইন্টারফেস (UI)। কিছু অ্যাপের গ্রাফিক্যাল ইউজার ইন্টারফেস (GUI) থাকে যেখানে আপনি বোতাম ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন নির্বাচন করেন এবং আরও অনেক কিছু করেন।

### কনসোল এবং UI অ্যাপের সীমাবদ্ধতা

একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমাবদ্ধ**। আপনি যে কোনও কমান্ড টাইপ করতে পারবেন না, শুধুমাত্র সেই কমান্ডগুলি যা অ্যাপ সমর্থন করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সমর্থন করে, কিন্তু ডিফল্টরূপে অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি আরো ভাষা সমর্থন যোগ করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে টেক্সট জেনারেশন অ্যাপ কীভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে, আপনার আরও বেশি নমনীয়তা থাকে, আপনি কমান্ডের একটি সেট বা একটি নির্দিষ্ট ইনপুট ভাষায় সীমাবদ্ধ নন। এর বদলে, আপনি প্রাকৃতিক ভাষা ব্যবহার করে অ্যাপের সাথে ইন্টারঅ্যাক্ট করতে পারেন। আরেকটি সুবিধা হল আপনি ইতোমধ্যে এমন একটি ডেটা উৎসের সাথে ইন্টারঅ্যাক্ট করছেন যা বিস্তৃত তথ্যভাণ্ডারের উপর প্রশিক্ষিত, যেখানে একটি প্রচলিত অ্যাপ ডাটাবেজে যা আছে তার মধ্যে সীমাবদ্ধ থাকতে পারে।

### আমি টেক্সট জেনারেশন অ্যাপ দিয়ে কী তৈরি করতে পারি?

অনেক কিছু তৈরি করতে পারেন। যেমন:

- **একটি চ্যাটবট**। একটি চ্যাটবট যা বিষয় সম্পর্কে প্রশ্নের উত্তর দেয়, যেমন আপনার কোম্পানি এবং এর পণ্যগুলি, একটি ভাল মিল হতে পারে।
- **সহকারী**। LLM গুলো টেক্সট সারাংশ তৈরি, টেক্সট থেকে অন্তর্দৃষ্টি পাওয়া, রিজুমে বা টেক্সট উৎপাদনের মতো কাজগুলোতে দারুন দক্ষ।
- **কোড সহকারী**। আপনি যে ভাষা মডেল ব্যবহার করেন তার উপর নির্ভর করে, আপনি একটি কোড সহকারী তৈরি করতে পারেন যা কোড লেখায় সাহায্য করে। উদাহরণস্বরূপ, আপনি GitHub Copilot এবং ChatGPT এর মতো পণ্য ব্যবহার করতে পারেন কোড লেখায় সাহায্য করার জন্য।

## কীভাবে শুরু করব?

সাধারণত, আপনাকে একটি LLM এর সাথে ইন্টিগ্রেট করার উপায় খুঁজে পেতে হবে, যা সাধারণত নিম্নলিখিত দুটি পন্থার মধ্যে একটি:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পটসহ ওয়েব অনুরোধ তৈরি করেন এবং তৈরি করা টেক্সট ফিরে পান।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলো API কলগুলোকে আবৃত করে এবং ব্যবহার সহজ করে তোলে।

## লাইব্রেরি/SDK গুলো

LLM এর সাথে কাজ করার জন্য কয়েকটি পরিচিত লাইব্রেরি রয়েছে, যেমন:

- **openai**, এই লাইব্রেরি আপনার মডেলের সাথে সংযোগ স্থাপন এবং প্রম্পট পাঠানো সহজ করে।

তারপর কিছু উচ্চস্তরের লাইব্রেরি রয়েছে, যেমন:

- **Langchain**। Langchain বিখ্যাত এবং Python সমর্থন করে।
- **Semantic Kernel**। Semantic Kernel মাইক্রোসফটের একটি লাইব্রেরি যা C#, Python, এবং Java ভাষাগুলিকে সমর্থন করে।

### চলুন আমাদের নিজের IDE তে মডেলটি ব্যবহার করি

এখানে দুটি অপশন:

2. **VS Code (অথবা যেকোনো পছন্দের IDE)** - এর জন্য আপনার Foundry প্রকল্প থেকে **endpoint** এবং **API key** প্রয়োজন হবে


যেকোনো উপায়ে, এই মানগুলা আপনি আপনার Foundry প্রকল্পের **Overview** পেজে পাবেন।

### 1. Codespaces 

* একটি নতুন codespace তৈরি করুন (অথবা একটি বিদ্যমান ব্যবহার করুন)
* VS Code ব্রাউজারে খুলবে যেখানে একাধিক ভাষায় নমুনা নোটবুক থাকবে যা আপনি চেষ্টা করতে পারবেন
* নমুনা ```./githubmodels-app.py``` রান করুন।

> দ্রষ্টব্য: Codespaces এ, আপনার Foundry endpoint এবং key কোডস্পেস সিক্রেট হিসেবে সংরক্ষণ করুন যাতে এগুলো লোকালি সেট করতে না হয়।

## অনুশীলন - একটি রেসিপি জেনারেটর তৈরি করুন

এখন যেহেতু আমরা একটি দৃশ্যপরিকর খেলার মাধ্যমে দেখিয়েছি, চলুন সেই অনুযায়ী কোড লিখি। এর জন্য নিচের ধাপগুলো অনুসরণ করুন:

১। বিদ্যমান ফাইলটি শুরু করার পয়েন্ট হিসেবে ব্যবহার করুন
১। একটি `prompt` ভ্যারিয়েবল তৈরি করুন এবং নমুনা কোড নিচের মতো পরিবর্তন করুন:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

আমাদের এখন একটি কাজ করার যোগ্য অ্যাপ্লিকেশন আছে যা রেসিপি তৈরি করতে পারে এবং এটি নমনীয় কারণ এটি ব্যবহারকারীর ইনপুটের উপর নির্ভর করে, শুধুমাত্র রেসিপির সংখ্যা নয় বরং ব্যবহৃত উপকরণ সম্পর্কেও।

আরও উন্নতির জন্য, আমরা নিম্নলিখিত যোগ করতে চাই:

- **উপকরণ ফিল্টার করুন**। আমরা এমন উপকরণ ফিল্টার করতে পারি যা আমরা পছন্দ করি না অথবা যার প্রতি অ্যালার্জি আছে। এই পরিবর্তনটি করতে, আমরা আমাদের বিদ্যমান প্রম্পট সম্পাদনা করে শেষে একটি ফিল্টার শর্ত যোগ করতে পারি এভাবে:
