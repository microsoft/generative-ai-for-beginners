# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

আপনি এখন পর্যন্ত এই পাঠক্রমে দেখেছেন যে মূল ধারণাগুলি রয়েছে যেমন প্রম্পট এবং এমনকি একটি সম্পূর্ণ শাস্ত্র "প্রম্পট ইঞ্জিনিয়ারিং" নামে। অনেক সরঞ্জাম যেমন ChatGPT, Office 365, Microsoft Power Platform ইত্যাদি আপনাকে প্রম্পট ব্যবহার করে কিছু সম্পন্ন করতে সাহায্য করে।

একটি অ্যাপে এমন অভিজ্ঞতা যোগ করতে, আপনাকে প্রম্পট, কমপ্লিশনস-এর মতো ধারণাগুলি বুঝতে হবে এবং কাজ করার জন্য একটি লাইব্রেরি নির্বাচন করতে হবে। ঠিক এটাই আপনি এই অধ্যায়ে শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলো সম্পর্কে শিখবেন।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, টেম্পারেচার, এবং টোকেনের মতো ধারণাগুলো ব্যবহার করে কীভাবে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করা যায় তা বুঝবেন।

## শেখার লক্ষ্য

এই পাঠের শেষে আপনি সক্ষম হবেন:

- কি একটি টেক্সট জেনারেশন অ্যাপ তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপকে বেশি বা কম টোকেন ব্যবহার করার জন্য কনফিগার করতে এবং আউটপুটের বৈচিত্র্যের জন্য টেম্পারেচার পরিবর্তন করতে।

## টেক্সট জেনারেশন অ্যাপ কি?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন তখন সেটিতে কিছু ধরণের ইন্টারফেস থাকে যেমন নিচের মত:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপ হলো সাধারণ অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং এটি একটি কাজ সম্পাদন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ব্যবহারকারী ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিক্যাল ইউজার ইন্টারফেস (GUI) থাকে যেখানে আপনি বাটনে ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন নির্বাচন করেন ইত্যাদি।

### কনসোল এবং UI অ্যাপগুলোর সীমাবদ্ধতা

এটি একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমাবদ্ধ।** আপনি যেকোনো কমান্ড টাইপ করতে পারবেন না, শুধুমাত্র সেই কমান্ডগুলি যা অ্যাপ সমর্থন করে।
- **ভাষা-নির্ধারিত।** কিছু অ্যাপ অনেক ভাষা সমর্থন করে, কিন্তু ডিফল্টরূপে অ্যাপ একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি আরো ভাষা যোগ করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে টেক্সট জেনারেশন অ্যাপ কিভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে, আপনার আরো নমনীয়তা থাকে, আপনি কমান্ড বা নির্দিষ্ট ইনপুট ভাষার সীমাবদ্ধ নয়। পরিবর্তে, আপনি প্রাকৃতিক ভাষা ব্যবহার করে অ্যাপের সাথে যোগাযোগ করতে পারেন। আরেকটি সুবিধা হলো, আপনি ইতোমধ্যে এমন একটি ডেটা উৎসের সাথে যোগাযোগ করছেন যা একটি ব্যাপক তথ্যভাণ্ডারে প্রশিক্ষিত, যেখানে একটি প্রচলিত অ্যাপ ডেটাবেসে থাকা তথ্য সীমাবদ্ধ থাকতে পারে।

### টেক্সট জেনারেশন অ্যাপ দিয়ে কি কি তৈরি করতে পারি?

অনেক কিছু তৈরি করা যায়। উদাহরণস্বরূপ:

- **একটি চ্যাটবট।** একটি চ্যাটবট যা প্রশ্নের উত্তর দেয়, যেমন আপনার কোম্পানি এবং এর পণ্য সম্পর্কে, সেটি একটি ভাল ম্যাচ হতে পারে।
- **সহায়ক।** LLM গুলি যেমন টেক্সট সারসংক্ষেপ করা, টেক্সট থেকে ইনসাইট সংগ্রহ করা, রিজিউমে তৈরি করা ইত্যাদিতে দারুন কাজ করে।
- **কোড সহায়ক।** আপনি যে ভাষার মডেল ব্যবহার করেন তার উপর নির্ভর করে, আপনি এমন একটি কোড সহায়ক তৈরি করতে পারেন যা আপনাকে কোড লেখায় সাহায্য করে। উদাহরণস্বরূপ, আপনি GitHub Copilot বা ChatGPT ব্যবহার করতে পারেন কোড লেখায় সাহায্য করার জন্য।

## কিভাবে শুরু করব?

প্রথমে আপনাকে একটি LLM-এর সাথে ইন্টিগ্রেট করার উপায় খুঁজে বের করতে হবে যা সাধারণত নিম্নলিখিত দুটি পদ্ধতি অন্তর্ভুক্ত:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পটসহ ওয়েব রিকোয়েস্ট তৈরি করেন এবং উত্পন্ন টেক্সট পান।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলো API কলগুলোকে এনক্যাপসুলেট করে এবং ব্যবহার সহজ করে তোলে।

## লাইব্রেরি/SDKs

LLM-গুলোর সাথে কাজ করার জন্য কয়েকটি পরিচিত লাইব্রেরি রয়েছে যেমন:

- **openai**, এই লাইব্রেরি আপনার মডেলের সঙ্গে সহজে সংযোগ স্থাপন করতে এবং প্রম্পট পাঠাতে সহায়তা করে।

এছাড়া কিছু লাইব্রেরি রয়েছে যা উচ্চ স্তরে কাজ করে, যেমন:

- **Langchain**। Langchain জনপ্রিয় এবং পাইথন সমর্থন করে।
- **Semantic Kernel**। Semantic Kernel মাইক্রোসফটের একটি লাইব্রেরি যা C#, Python, এবং Java ভাষাসমূহকে সমর্থন করে।

### ১. Codespaces 

* একটি নতুন codespace তৈরি করুন (অথবা একটি বিদ্যমান ব্যবহার করুন)
* VS Code আপনার ব্রাউজারে খুলবে এবং আপনি একাধিক ভাষার নমুনা নোটবুক চেষ্টা করতে পারবেন
* নমুনা রান করুন ```./githubmodels-app.py```

> দ্রষ্টব্য: Codespaces-এ, আপনার Foundry endpoint ও কী Codespaces secrets হিসেবে সংরক্ষণ করুন যাতে লোকালি সেট করতে না হয়।

## ব্যায়াম - একটি রেসিপি জেনারেটর তৈরি করুন

এখন যেহেতু আমরা একটি দৃশ্যপট শেয়ার করেছি, চলুন প্রদর্শিত দৃশ্যপটের সাথে মিল রেখে কোড লিখি। তা করার জন্য, নিম্নলিখিত ধাপগুলি অনুসরণ করুন:

1. বর্তমান ফাইলটি শুরু বিন্দু হিসেবে ব্যবহার করুন
1. একটি `prompt` ভেরিয়েবল তৈরি করুন এবং নমুনা কোড নিচের মতো পরিবর্তন করুন:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

এখন যদি আপনি কোডটি চালান, তাহলে আপনি এমন একটি আউটপুট দেখতে পাবেন যা এরকম হবে:

দারুণ, চলুন দেখি আমরা কীভাবে বিষয়গুলো উন্নত করতে পারি। বিষয়গুলো উন্নত করতে, আমরা চাই কোডটি নমনীয় হোক, যাতে উপাদান এবং রেসিপির সংখ্যা উন্নত এবং পরিবর্তন করা যায়।


1. আসুন নিম্নলিখিতভাবে কোডটি পরিবর্তন করি:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

এখন আমাদের একটি কাজ করা অ্যাপ আছে যা রেসিপি তৈরি করতে সক্ষম এবং এটি নমনীয় কারণ এটি ব্যবহারকারীর ইনপুটের উপর ভিত্তি করে, রেসিপির সংখ্যাও এবং ব্যবহৃত উপাদানও।

আরও উন্নতির জন্য আমরা নিম্নলিখিত যোগ করতে চাই:

- **উপাদানগুলি ফিল্টার করুন**। আমরা এমন উপাদানগুলি ফিল্টার করতে চাই যেগুলি আমরা পছন্দ করি না বা যেগুলির প্রতি আমাদের অ্যালার্জি আছে। এই পরিবর্তনটি সম্পন্ন করতে আমরা আমাদের বিদ্যমান প্রম্পট সম্পাদনা করে এর শেষে একটি ফিল্টার শর্ত যোগ করতে পারি যেমন:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    উপরোক্ত অংশে, আমরা প্রম্পটের শেষে `{filter}` যোগ করেছি এবং ব্যবহারকারীর কাছ থেকে ফিল্টার মান গ্রহণ করেছি।

    প্রোগ্রাম চালানোর একটি উদাহরণ ইনপুট এখন এই রকম হতে পারে:

- **একটি শপিং লিস্ট তৈরি করুন**। আমরা একটি শপিং লিস্ট তৈরি করতে চাই, যেটা বিবেচনা করবে আমাদের বাড়িতে ইতিমধ্যেই যা আছে।

    এই ফাংশনালিটির জন্য, আমরা সবকিছু এক প্রম্পটে সমাধান করার চেষ্টা করতে পারি অথবা দুইটি প্রম্পটে ভাগ করতে পারি। পরের পদ্ধতি চেষ্টা করি। এখানে আমরা একটি অতিরিক্ত প্রম্পট যোগ করার পরামর্শ দিচ্ছি, তবে তার জন্য প্রথম প্রম্পটের ফলাফল দ্বিতীয় প্রম্পটের প্রসঙ্গে যুক্ত করতে হবে।

    কোডের সেই অংশ খুঁজে বের করুন যা প্রথম প্রম্পটের ফলাফল প্রিন্ট করে এবং নিচের কোডটি তার নিচে যোগ করুন:

এখানে কিছু পরামর্শ দেওয়া হলো:

- ইতিহাস বট, ইতিহাসকে জীবিত করুন, বটকে নির্দেশ দিন একটি নির্দিষ্ট ঐতিহাসিক চরিত্র হিসেবে ভূমিকা পালন করতে এবং তার জীবন ও সময় সম্পর্কে প্রশ্ন করুন।
