# টেক্সট জেনারেশন অ্যাপস তৈরি করুন

আপনি যে পর্যন্ত এই পাঠক্রমটি অনুসরণ করেছেন, সেখানে মূল ধারণাগুলি যেমন প্রম্পট এবং এমনকি একটি সম্পূর্ণ শাখা "প্রম্পট ইঞ্জিনিয়ারিং" রয়েছে। অনেক সরঞ্জাম যার সাথে আপনি ইন্টারঅ্যাক্ট করতে পারেন যেমন ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক কিছু, যা আপনাকে কিছু অর্জন করতে প্রম্পট ব্যবহার করতে সহায়তা করে।

এমন একটি অভিজ্ঞতা এক অ্যাপে যোগ করার জন্য, আপনাকে প্রম্পট, সম্পূর্ণতা (completions) ইত্যাদি ধারণাগুলি বুঝতে হবে এবং কাজ করার জন্য একটি লাইব্রেরি নির্বাচন করতে হবে। সেটাই আপনি এই অধ্যায়ে শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলি সম্পর্কে শিখবেন।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, টেম্পারেচার এবং টোকেনের মতো ধারণাগুলি কীভাবে টেক্সট জেনারেশন অ্যাপ তৈরিতে ব্যবহার করবেন তা বুঝবেন।

## শেখার লক্ষ্য

এই পাঠ শেষ হলে, আপনি সক্ষম হবেন:

- টেক্সট জেনারেশন অ্যাপ কী তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপকে কনফিগার করতে যাতে বেশি বা কম টোকেন ব্যবহার করা যায় এবং তাপমাত্রাও পরিবর্তন করা যায়, ভিন্ন আউটপুটের জন্য।

## টেক্সট জেনারেশন অ্যাপ কী?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, তখন সেটির কিছু ধরনের ইন্টারফেস থাকে যেমন:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপস এমন অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং এটি একটি কাজ সম্পাদন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ব্যবহারকারী ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিকাল ইউজার ইন্টারফেস (GUI) থাকে, যেখানে আপনি বাটনে ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন বাছাই করেন এবং আরও অনেক কিছু।

### কনসোল এবং UI অ্যাপ সীমাবদ্ধ

এর সাথে একটি কমান্ড-ভিত্তিক অ্যাপ তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **এটি সীমাবদ্ধ**। আপনি যেকোনো কমান্ড টাইপ করতে পারবেন না, শুধুমাত্র যেগুলো অ্যাপ সাপোর্ট করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সাপোর্ট করে, কিন্তু ডিফল্টরূপে অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি অতিরিক্ত ভাষা সাপোর্ট যোগ করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে টেক্সট জেনারেশন অ্যাপ কীভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে, আপনার বেশি নমনীয়তা থাকে, আপনি একটি নির্দিষ্ট কমান্ড সেট বা নির্দিষ্ট ইনপুট ভাষায় সীমাবদ্ধ নন। পরিবর্তে, আপনি ন্যাচারাল ভাষা ব্যবহার করে অ্যাপের সাথে ইন্টার‍্যাক্ট করতে পারেন। আরেকটি সুবিধা হচ্ছে আপনি এমন একটি ডেটা সোর্সের সাথে যোগাযোগ করছেন যা বিশাল তথ্যভিত্তিক কর্পাসে প্রশিক্ষিত, যেখানে ঐতিহ্যবাহী অ্যাপ সাধারণত একটি ডাটাবেসে থাকা তথ্যের মধ্যেই সীমাবদ্ধ থাকে।

### টেক্সট জেনারেশন অ্যাপ দিয়ে আমি কী তৈরি করতে পারি?

অনেক কিছু আপনি তৈরি করতে পারেন। উদাহরণস্বরূপ:

- **একটি চ্যাটবট**। একটি চ্যাটবট যেটি বিষয়বস্তু সম্পর্কে, যেমন আপনার কোম্পানি এবং এর পণ্য সম্পর্কে প্রশ্নের উত্তর দেয়, ভালো মানানসই হতে পারে।
- **সহায়ক**। LLM গুলো টেক্সট সারাংশ তৈরি, টেক্সট থেকে ইনসাইট পাওয়া, রেজুমে তৈরি করা ইত্যাদিতে খুব ভাল।
- **কোড সহকারী**। আপনি যে ভাষা মডেল ব্যবহার করেন তার উপরে নির্ভর করে, আপনি এমন একটি কোড সহকারী তৈরি করতে পারেন যা কোড লেখতে সাহায্য করে। উদাহরণস্বরূপ, আপনি GitHub Copilot অথবা ChatGPT ব্যবহার করে কোড লেখার সহায়তা নিতে পারেন।

## কীভাবে শুরু করবেন?

ভাল, আপনাকে একটি LLM এর সাথে সংযুক্তির কোনো উপায় খুঁজতে হবে যা সাধারণত নিম্নলিখিত দুইটি পদ্ধতি গ্রহণ করে:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পট নিয়ে ওয়েব রিকোয়েস্ট তৈরি করেন এবং জেনারেটেড টেক্সট ফিরে পান।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলো API কলগুলোকে ক্যাপসুলেট করে এবং ব্যবহারে সহজ করে তোলে।

## লাইব্রেরি/SDK

LLM এর সাথে কাজ করার জন্য কিছু পরিচিত লাইব্রেরি রয়েছে যেমন:

- **openai**, এই লাইব্রেরি আপনার মডেলের সাথে যুক্ত হওয়া এবং প্রম্পট পাঠানো সহজ করে তোলে।

এরপর উচ্চ স্তরের কাজ করার লাইব্রেরি রয়েছে:

- **Langchain**। Langchain পরিচিত এবং এটি পাইথন সাপোর্ট করে।
- **Semantic Kernel**। Semantic Kernel হলো মাইক্রোসফটের একটি লাইব্রেরি যা C#, পাইথন এবং জাভা ভাষাগুলোকে সাপোর্ট করে।

## প্রথম অ্যাপ openai ব্যবহার করে

চলুন দেখি কিভাবে আমরা আমাদের প্রথম অ্যাপ তৈরি করতে পারি, কোন লাইব্রেরিগুলো প্রয়োজন, কতটা দরকার এবং অন্যান্য বিষয়।

OpenAI বা Azure OpenAI এর সাথে ইন্টার‍্যাক্ট করার অনেক লাইব্রেরি রয়েছে। C#, পাইথন, জাভাস্ক্রিপ্ট, জাভাসহ বিভিন্ন প্রোগ্রামিং ভাষাও ব্যবহার করা যায়।  
আমরা `openai` পাইথন লাইব্রেরি নির্বাচন করেছি, তাই এটি ইনস্টল করতে `pip` ব্যবহার করব।

```bash

```

যদি আপনি এই নোটবুকটি Codespaces বা Dev Container-এ না চালান, তাহলে আপনাকে আপনার মেশিনে [Python](https://www.python.org/) ইনস্টল করতেও হবে।

### একটি রিসোর্স তৈরি করুন এবং আপনার API কী অবস্থান করুন

যদি আপনি এখনও না করে থাকেন, তাহলে নিম্নলিখিত ধাপগুলি অনুসরণ করতে হবে:

- OpenAI তে একটি অ্যাকাউন্ট তৈরি করুন <https://platform.openai.com/signup>।
- এখন আপনার API কীগুলো পান <https://platform.openai.com/api-keys>।

>[!NOTE]
> আপনার API কী কোড থেকে আলাদা রাখা ভাল। আপনি পরিবেশ পরিবর্তনশীল ব্যবহার করে এটি করতে পারেন।
> - আপনার API কী `OPENAI_KEY` পরিবেশ পরিবর্তনশীল হিসেবে আপনার .env ফাইলে সেট করুন। যদি আপনি ইতিমধ্যে এই কোর্সের পূর্বের অনুশীলনগুলি করেছেন, তাহলে আপনি প্রস্তুত আছেন।
> - এটি গুরুত্বপূর্ণ যে API কী একবারই অ্যাক্সেসযোগ্য হবে। তাই নিশ্চিত করুন যে এটি সঠিকভাবে কপি করা হয়েছে। যদি এটি সঠিকভাবে কাজ না করে, কীটি মুছে নতুন কী জেনারেট করুন।


### OpenAI কনফিগারেশন সেটআপ

আপনি যদি OpenAI ব্যবহার করেন, তাহলে কনফিগারেশন সেটআপ করার পদ্ধতি হলো:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

উপরে আমরা নিম্নলিখিত বিষয়গুলো সেট করছি:

- `api_key`, এটি আপনার OpenAI ড্যাশবোর্ডে পাওয়া API কী।
- `deployment`, এটি আপনার GPT ভার্সন।

> [!NOTE]
> `os.environ` একটি ফাংশন যা পরিবেশ পরিবর্তনশীল পড়ে। আপনি এটি ব্যবহার করে `OPENAI_API_KEY` এর মতো পরিবেশ পরিবর্তনশীল পড়তে পারেন।

## টেক্সট জেনারেট করুন

টেক্সট জেনারেট করার পদ্ধতি হলো `responses.create` মেথড ব্যবহার করা। এখানে একটি উদাহরণ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

উপরের কোডে, আমরা একটি রেসপন্স অবজেক্ট তৈরি করি এবং ব্যবহারের জন্য মডেল ও প্রম্পট দিয়ে থাকি। তারপর আমরা তৈরি হওয়া টেক্সট প্রিন্ট করি।

### চ্যাট রেসপন্স

এখন পর্যন্ত, আপনি দেখেছেন কীভাবে একটি প্রম্পট থেকে টেক্সট তৈরি করা যায়। Responses API চ্যাটবটগুলোর জন্যও উপযুক্ত - আপনি `input` হিসেবে মেসেজের তালিকা প্রদান করেন কথোপকথন গড়ে তোলার জন্য। এখানে একটি উদাহরণ দেখুন:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

এই ফাংশনালিটির আরও বিস্তারিত একটি পরবর্তী অধ্যায়ে আসবে।

## বিভিন্ন ধরণের প্রম্পট, বিভিন্ন কাজের জন্য

এখন আপনি দেখেছেন কীভাবে একটি প্রম্পট ব্যবহার করে টেক্সট তৈরি করা যায়। আপনার কাছে এমন একটি প্রোগ্রামও আছে যা চালু আছে এবং আপনি সেটি পরিবর্তন করে বিভিন্ন ধরণের টেক্সট তৈরি করতে পারেন।

প্রম্পট বিভিন্ন ধরনের কাজের জন্য ব্যবহার করা যেতে পারে। উদাহরণস্বরূপ:

- **একটি ধরণের টেক্সট তৈরি করা**। যেমন, আপনি কবিতা, কুইজের জন্য প্রশ্ন ইত্যাদি তৈরি করতে পারেন।
- **তথ্য অনুসন্ধান করা**। আপনি প্রম্পট ব্যবহার করে তথ্য অনুসন্ধান করতে পারেন, যেমন 'ওয়েব ডেভেলপমেন্টে CORS এর অর্থ কী?' এই উদাহরণটি।
- **কোড তৈরি করা**। আপনি প্রম্পট ব্যবহার করে কোড তৈরি করতে পারেন, যেমন ইমেইল যাচাই করার জন্য একটি রেগুলার এক্সপ্রেশন তৈরি করা অথবা পুরো একটি প্রোগ্রাম, যেমন একটি ওয়েব অ্যাপ তৈরি করা।

## অনুশীলন - একটি রেসিপি তৈরি করার জেনারেটর তৈরি করা

এখন যেহেতু আমরা একটি পরিস্থিতি অনুশীলন করেছি, চল সেই প্রদর্শিত পরিস্থিতির সাথে মানানসই কোড লিখি। এর জন্য, নিচের ধাপগুলো অনুসরণ করুন:

1. বিদ্যমান *app.py* ফাইলটি শুরু করার জন্য ব্যবহার করুন
1. `prompt` ভেরিয়েবলটি খুঁজে বের করুন এবং এর কোড নিম্নরূপ পরিবর্তন করুন:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### ফিল্টার এবং শপিং লিস্ট যোগ করে উন্নতি করুন

এখন আমাদের কাছে একটি কাজ করা অ্যাপ রয়েছে যা রেসিপি তৈরি করতে সক্ষম এবং এটি নমনীয় কারণ এটি ব্যবহারকারীর ইনপুটের উপর নির্ভর করে, রেসিপির সংখ্যা এবং ব্যবহৃত উপাদানের উপর।

এটি আরও উন্নত করতে, আমরা নিম্নলিখিত যোগ করতে চাই:

- **উপাদান ফিল্টার করুন**। আমরা এমন উপাদান ফিল্টার করতে চাই যা আমরা পছন্দ করি না বা যার প্রতি অ্যালার্জি আছে। এই পরিবর্তনটি করতে, আমরা আমাদের বিদ্যমান প্রম্পট সম্পাদনা করে এর শেষে একটি ফিল্টার শর্ত যোগ করতে পারি এইভাবে:

    দেখা যাচ্ছে, যার মধ্যে দুধ আছে এমন রেসিপিগুলো ফিল্টার করা হয়েছে। কিন্তু যদি আপনি ল্যাকটোজ অ্যালার্জি থাকেন, আপনি চাইতে পারেন পনিরসহ রেসিপি গুলিও ফিল্টার করতে, সেজন্য এটা স্পষ্ট করা জরুরি।

- **শপিং লিস্ট তৈরি করুন**। আমরা একটি শপিং লিস্ট তৈরি করতে চাই, যা বিবেচনা করবে যে আমরা ঘরে কি কিছু আছে।

    এই ফাংশনালিটির জন্য, আমরা হয় একটি প্রম্পটে সব কিছু সমাধান করার চেষ্টা করতে পারি অথবা দুইটি প্রম্পটে ভাগ করতে পারি। আমরা দ্বিতীয় পদ্ধতি চেষ্টা করব। এখানে আমরা একটি অতিরিক্ত প্রম্পট যোগ করার পরামর্শ দিচ্ছি, কিন্তু তা কাজ করার জন্য, আমাদের প্রথম প্রম্পটের ফলাফল দ্বিতীয় প্রম্পটের প্রসঙ্গে যোগ করতে হবে।

    কোডের যে অংশ প্রথম প্রম্পট থেকে ফলাফল প্রিন্ট করে তা খুঁজে বের করুন এবং নিচের কোডটি সেখানে যুক্ত করুন:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # রেসপন্স প্রিন্ট করুন
    print("Shopping list:")
    print(response.output_text)
    ```

    নিম্নলিখিত বিষয়গুলি লক্ষ্য করুন:

    - আমরা নতুন একটি প্রম্পট তৈরি করছি যা প্রথম প্রম্পটের ফলাফল নতুন প্রম্পটের সাথে যুক্ত:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - আমরা নতুন অনুরোধ করছি, কিন্তু প্রথম প্রম্পটে ব্যবহৃত টোকেনের সংখ্যা বিবেচনায় রেখে, তাই এবার আমরা `max_output_tokens` 1200 দিয়েছি।

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        এই কোডটি চালালে এখন নিচের আউটপুট পাওয়া যাবে:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.
        
        -Apple and flour pancakes: 1 cup flour, 1/2 tsp baking powder, 1/2 tsp baking soda, 1/4 tsp salt, 1 tbsp sugar, 1 egg, 1 cup buttermilk or sour milk, 1/4 cup melted butter, 1 Granny Smith apple, peeled and grated
        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **টোকেন দৈর্ঘ্য নিয়ে একটি কথা**। আমাদের যতটুকু টোকেনের দরকার আছে তা বিবেচনা করা উচিত কারণ টোকেন ব্যয়বহুল, সুতরাং যেখানে সম্ভব, টোকেনের সংখ্যা কম রাখার চেষ্টা করা উচিত। উদাহরণস্বরূপ, আমরা কি প্রম্পটকে এমনভাবে সাজাতে পারি যাতে কম টোকেন ব্যবহার হয়?

   টোকেন পরিবর্তন করতে, আপনি `max_output_tokens` প্যারামিটার ব্যবহার করতে পারেন। যেমন, আপনি যদি ১০০ টোকেন ব্যবহার করতে চান, তাহলে করবেন:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **তাপমাত্রা নিয়ে পরীক্ষা**। তাপমাত্রা এমন একটি বিষয় যা আমরা এখন পর্যন্ত উল্লেখ করিনি কিন্তু আমাদের প্রোগ্রাম কিভাবে কাজ করে তার জন্য গুরুত্বপূর্ণ। তাপমাত্রার মান যত বেশি হয় আউটপুট তত বেশি এলোমেলো হবে। অপরদিকে তাপমাত্রার মান যত কম হবে অউটপুট তত বেশি পূর্বানুমানযোগ্য হবে। বিবেচনা করুন আপনি কি আউটপুটে বৈচিত্র্য চান কি না।

   তাপমাত্রা পরিবর্তন করতে, আপনি `temperature` প্যারামিটার ব্যবহার করতে পারেন। যেমন, আপনি যদি ০.৫ তাপমাত্রা ব্যবহার করতে চান, তাহলে করবেন:

    ```python

    ```

   > লক্ষণীয়, ১.০ এর কাছে যত বেশি থাকবে ততই আউটপুট বৈচিত্র্যময় হবে।
