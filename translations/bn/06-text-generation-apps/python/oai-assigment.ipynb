# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

এই কারিকুলামে এখন পর্যন্ত আপনি দেখেছেন যে প্রম্পটের মতো কিছু মূল ধারণা আছে এবং এমনকি "প্রম্পট ইঞ্জিনিয়ারিং" নামে একটি সম্পূর্ণ শাখাও আছে। ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক টুলে আপনি প্রম্পট ব্যবহার করে কাজ করতে পারেন।

আপনি যদি আপনার অ্যাপে এমন অভিজ্ঞতা যোগ করতে চান, তাহলে আপনাকে প্রম্পট, কমপ্লিশন এবং কোন লাইব্রেরি ব্যবহার করবেন তা জানতে হবে। এই অধ্যায়ে আপনি ঠিক এগুলোই শিখবেন।

## পরিচিতি

এই অধ্যায়ে, আপনি শিখবেন:

- openai লাইব্রেরি এবং এর মূল ধারণা সম্পর্কে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- প্রম্পট, টেম্পারেচার, এবং টোকেনের মতো ধারণা ব্যবহার করে কিভাবে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে হয়।

## শেখার লক্ষ্য

এই পাঠ শেষে, আপনি পারবেন:

- টেক্সট জেনারেশন অ্যাপ কী তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপে কত টোকেন ব্যবহার হবে এবং টেম্পারেচার পরিবর্তন করে আউটপুটে বৈচিত্র্য আনতে।

## টেক্সট জেনারেশন অ্যাপ কী?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, তার কিছু ইন্টারফেস থাকে যেমন:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপগুলো সাধারণত এমন অ্যাপ যেখানে আপনি একটি কমান্ড লিখে কাজ করান। যেমন, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ইউজার ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিকাল ইউজার ইন্টারফেস (GUI) থাকে যেখানে আপনি বাটনে ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন সিলেক্ট করেন ইত্যাদি।

### কনসোল ও UI অ্যাপের সীমাবদ্ধতা

একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি কমান্ড লিখেন:

- **এটি সীমিত**। আপনি যেকোনো কমান্ড লিখতে পারবেন না, শুধু যেগুলো অ্যাপ সাপোর্ট করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সাপোর্ট করে, কিন্তু সাধারণত অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি, যদিও আপনি আরও ভাষা যোগ করতে পারেন।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে টেক্সট জেনারেশন অ্যাপ কীভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে আপনি আরও বেশি ফ্লেক্সিবিলিটি পান, আপনি নির্দিষ্ট কিছু কমান্ড বা ভাষার মধ্যে সীমাবদ্ধ নন। বরং, আপনি প্রাকৃতিক ভাষা ব্যবহার করে অ্যাপের সাথে কথা বলতে পারেন। আরেকটি সুবিধা হলো, আপনি এমন একটি ডেটা সোর্সের সাথে কাজ করছেন যা বিশাল তথ্যভাণ্ডারে প্রশিক্ষিত, যেখানে সাধারণ অ্যাপ কেবল ডাটাবেসের তথ্যেই সীমাবদ্ধ।

### টেক্সট জেনারেশন অ্যাপ দিয়ে কী বানানো যায়?

অনেক কিছু বানানো যায়। যেমন:

- **চ্যাটবট**। চ্যাটবট যেটি আপনার কোম্পানি ও পণ্যের বিষয়ে প্রশ্নের উত্তর দিতে পারে।
- **সহায়ক**। LLM গুলো টেক্সট সারাংশ করা, টেক্সট থেকে ইনসাইট বের করা, সিভি বা অন্য টেক্সট তৈরি করার মতো কাজে দারুণ।
- **কোড অ্যাসিস্ট্যান্ট**। আপনি কোন ভাষার মডেল ব্যবহার করছেন তার ওপর নির্ভর করে, আপনি এমন কোড অ্যাসিস্ট্যান্ট বানাতে পারেন যা কোড লিখতে সাহায্য করে। যেমন, GitHub Copilot বা ChatGPT ব্যবহার করে কোড লেখার সহায়তা নিতে পারেন।

## কিভাবে শুরু করব?

আপনাকে LLM-এর সাথে সংযোগ করার উপায় খুঁজতে হবে, সাধারণত দুটি উপায়ে:

- API ব্যবহার করুন। এখানে আপনি ওয়েব রিকোয়েস্ট তৈরি করেন প্রম্পটসহ এবং জেনারেটেড টেক্সট ফেরত পান।
- লাইব্রেরি ব্যবহার করুন। লাইব্রেরি API কলগুলো সহজ করে তোলে।

## লাইব্রেরি/SDK

LLM নিয়ে কাজ করার জন্য কিছু পরিচিত লাইব্রেরি আছে যেমন:

- **openai**, এই লাইব্রেরি দিয়ে সহজেই আপনার মডেলে সংযোগ করে প্রম্পট পাঠানো যায়।

এছাড়াও কিছু উচ্চ-স্তরের লাইব্রেরি আছে যেমন:

- **Langchain**। Langchain বেশ পরিচিত এবং Python সাপোর্ট করে।
- **Semantic Kernel**। Semantic Kernel হলো Microsoft-এর একটি লাইব্রেরি, যা C#, Python, এবং Java সাপোর্ট করে।

আমরা এখানে `openai` Python লাইব্রেরি ব্যবহার করব, তাই `pip` দিয়ে এটি ইনস্টল করব।

### রিসোর্স তৈরি করুন এবং API কী খুঁজে নিন

যদি আপনি আগে না করে থাকেন, তাহলে নিচের ধাপগুলো অনুসরণ করুন:

- OpenAI-তে একটি অ্যাকাউন্ট তৈরি করুন <https://platform.openai.com/signup>।
- এখন, আপনার API কী নিন <https://platform.openai.com/api-keys>।

>[!NOTE]
> আপনার API কী কোড থেকে আলাদা রাখা ভালো। আপনি environment variable ব্যবহার করে এটি করতে পারেন।
> - .env ফাইলে `OPENAI_KEY` environment variable-এ আপনার API কী সেট করুন। আপনি যদি এই কোর্সের আগের এক্সারসাইজগুলো শেষ করে থাকেন, তাহলে আপনি প্রস্তুত।
> - মনে রাখবেন, API কী কেবল একবারই দেখা যাবে। তাই সঠিকভাবে কপি হয়েছে কিনা যাচাই করুন। যদি কাজ না করে, তাহলে কী ডিলিট করে নতুন কী তৈরি করুন।

### OpenAI কনফিগারেশন সেটআপ

আপনি যদি OpenAI ব্যবহার করেন, তাহলে কনফিগারেশন এভাবে সেট করুন:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

উপরের কোডে আমরা নিচেরগুলো সেট করেছি:

- `api_key`, এটি OpenAI ড্যাশবোর্ড থেকে পাওয়া আপনার API কী।
- `deployment`, এটি আপনার GPT ভার্সন।

> [!NOTE]
> `os.environ` হলো একটি ফাংশন যা environment variable পড়ে। আপনি এটি দিয়ে `OPENAI_API_KEY` এর মতো environment variable পড়তে পারেন।

## টেক্সট জেনারেট করুন

টেক্সট জেনারেট করার জন্য `chat.completion` ক্লাস ব্যবহার করতে হয়। উদাহরণ:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

উপরের কোডে, আমরা একটি completion অবজেক্ট তৈরি করেছি এবং কোন মডেল ব্যবহার করব ও প্রম্পট পাঠিয়েছি। তারপর জেনারেটেড টেক্সট প্রিন্ট করেছি।

### চ্যাট কমপ্লিশন

এখন পর্যন্ত আপনি দেখেছেন কিভাবে `Completion` ব্যবহার করে টেক্সট জেনারেট করা যায়। কিন্তু আরেকটি ক্লাস আছে `ChatCompletion` নামে, যা চ্যাটবটের জন্য বেশি উপযোগী। উদাহরণ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

এই ফিচার নিয়ে পরবর্তী অধ্যায়ে আরও আলোচনা করা হবে।

## অনুশীলন - আপনার প্রথম টেক্সট জেনারেশন অ্যাপ

এখন যেহেতু আমরা OpenAI সার্ভিস সেটআপ ও কনফিগার করা শিখেছি, এবার আপনার প্রথম টেক্সট জেনারেশন অ্যাপ তৈরি করার পালা। অ্যাপ তৈরি করতে নিচের ধাপগুলো অনুসরণ করুন:


১. একটি ভার্চুয়াল এনভায়রনমেন্ট তৈরি করুন এবং openai ইনস্টল করুন:

  > [!NOTE] এই ধাপটি প্রয়োজন নেই যদি আপনি এই নোটবুকটি Codespaces-এ বা Devcontainer-এর মধ্যে চালান


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> আপনি যদি Windows ব্যবহার করেন তাহলে `source venv/bin/activate` এর পরিবর্তে `venv\Scripts\activate` টাইপ করুন।

> [!NOTE]
> আপনার OpenAI কী খুঁজে পেতে https://platform.openai.com/settings/organization/api-keys এ যান এবং `API keys` খুঁজুন। আপনি সেখানে একটি নতুন কী তৈরি করতে পারেন এবং সাথে সাথে তার মান কপি করতে পারবেন।


১. একটি *app.py* ফাইল তৈরি করুন এবং এতে নিম্নলিখিত কোড দিন:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## বিভিন্ন ধরণের প্রম্পট, বিভিন্ন কাজে

এখন আপনি দেখেছেন কিভাবে একটি প্রম্পট ব্যবহার করে টেক্সট তৈরি করা যায়। আপনার কাছে এমন একটি প্রোগ্রামও আছে, যেটা আপনি পরিবর্তন ও সংশোধন করে বিভিন্ন ধরণের টেক্সট তৈরি করতে পারেন।

প্রম্পট নানা কাজে ব্যবহার করা যায়। যেমন:

- **নির্দিষ্ট ধরণের টেক্সট তৈরি করা**। উদাহরণস্বরূপ, আপনি একটি কবিতা, কুইজের জন্য প্রশ্ন ইত্যাদি তৈরি করতে পারেন।
- **তথ্য খোঁজা**। আপনি প্রম্পট ব্যবহার করে তথ্য খুঁজতে পারেন, যেমন: 'ওয়েব ডেভেলপমেন্টে CORS মানে কী?'।

## অনুশীলন - একটি রেসিপি জেনারেটর তৈরি করুন

এখন যেহেতু আমরা একটি পরিস্থিতি দেখেছি, চলুন সেই পরিস্থিতির সাথে মিল রেখে কোড লিখি। এজন্য নিচের ধাপগুলো অনুসরণ করুন:

1. বিদ্যমান *app.py* ফাইলটি শুরু করার জন্য ব্যবহার করুন
1. `prompt` ভ্যারিয়েবলটি খুঁজে বের করুন এবং তার কোড নিচের মতো পরিবর্তন করুন:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=100)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, temperature=0.5)
    ```
