# টেক্সট জেনারেশন অ্যাপ তৈরি করুন

এই কারিকুলাম জুড়ে আপনি ইতোমধ্যেই দেখেছেন যে প্রম্পট এবং এমনকি একটি পুরো শৃঙ্খলা রয়েছে যার নাম "প্রম্পট ইঞ্জিনিয়ারিং"। অনেক সরঞ্জাম যেমন ChatGPT, Office 365, Microsoft Power Platform এবং আরও অনেক কিছু, আপনাকে কোন কিছু সম্পাদনের জন্য প্রম্পট ব্যবহার করতে সাহায্য করে।

একটি অ্যাপে এমন অভিজ্ঞতা যোগ করার জন্য, আপনাকে প্রম্পট, কমপ্লিশনসের মতো ধারণাগুলি বুঝতে হবে এবং কাজ করার জন্য একটি লাইব্রেরি নির্বাচন করতে হবে। এটাই এই অধ্যায়ে আপনি শিখবেন।

## ভূমিকা

এই অধ্যায়ে, আপনি:

- openai লাইব্রেরি এবং এর মূল ধারণাগুলো সম্পর্কে শিখবেন।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করবেন।
- প্রম্পট, টেম্পারেচার, এবং টোকেনস এর মতো ধারণাগুলো কীভাবে ব্যবহার করতে হয় তা শিখবেন।

## শেখার লক্ষ্য

এই পাঠের শেষে, আপনি সক্ষম হবেন:

- একটি টেক্সট জেনারেশন অ্যাপ কী তা ব্যাখ্যা করতে।
- openai ব্যবহার করে একটি টেক্সট জেনারেশন অ্যাপ তৈরি করতে।
- আপনার অ্যাপ কনফিগার করে বেশি বা কম টোকেন ব্যবহার করতে এবং টেম্পারেচার পরিবর্তন করতে, যাতে ভিন্ন ধরনের আউটপুট পাওয়া যায়।

## টেক্সট জেনারেশন অ্যাপ কী?

সাধারণত যখন আপনি একটি অ্যাপ তৈরি করেন, এতে কিছু ধরণের ইন্টারফেস থাকে যেমন নিচে দেওয়া হয়েছে:

- কমান্ড-ভিত্তিক। কনসোল অ্যাপগুলি হল এমন অ্যাপ যেখানে আপনি একটি কমান্ড টাইপ করেন এবং অ্যাপ কোনো কাজ সম্পাদন করে। উদাহরণস্বরূপ, `git` একটি কমান্ড-ভিত্তিক অ্যাপ।
- ইউজার ইন্টারফেস (UI)। কিছু অ্যাপে গ্রাফিকাল ইউজার ইন্টারফেস (GUIs) থাকে যেখানে আপনি বোতাম ক্লিক করেন, টেক্সট ইনপুট দেন, অপশন নির্বাচন করেন ইত্যাদি।

### কনসোল এবং UI অ্যাপের সীমাবদ্ধতা

একটি কমান্ড-ভিত্তিক অ্যাপের সাথে তুলনা করুন যেখানে আপনি একটি কমান্ড টাইপ করেন:

- **সীমাবদ্ধ**। আপনি যে কোনো কমান্ড টাইপ করতে পারবেন না, শুধুমাত্র যেগুলো অ্যাপ সমর্থন করে।
- **ভাষা নির্দিষ্ট**। কিছু অ্যাপ অনেক ভাষা সমর্থন করে, তবে ডিফল্ট হিসেবে অ্যাপটি একটি নির্দিষ্ট ভাষার জন্য তৈরি করা হয়, যদিও আরও ভাষা যোগ করা যায়।

### টেক্সট জেনারেশন অ্যাপের সুবিধা

তাহলে, টেক্সট জেনারেশন অ্যাপ কীভাবে আলাদা?

একটি টেক্সট জেনারেশন অ্যাপে, আপনার অনেক বেশি নমনীয়তা থাকে, আপনি নির্দিষ্ট কমান্ড বা নির্দিষ্ট ইনপুট ভাষায় সীমাবদ্ধ নন। পরিবর্তে, আপনি প্রাকৃতিক ভাষা ব্যবহার করে অ্যাপের সাথে ইন্টারঅ্যাক্ট করতে পারেন। আরেকটি সুবিধা হলো আপনি এমন একটি ডেটা সোর্সের সাথে ইন্টারঅ্যাক্ট করছেন যা ব্যাপক তথ্যভাণ্ডারে প্রশিক্ষিত, যেখানে একটি প্রচলিত অ্যাপ ডাটাবেসে থাকা তথ্যের মধ্যে সীমাবদ্ধ থাকতে পারে।

### আমি কি কি বানাতে পারি একটি টেক্সট জেনারেশন অ্যাপ দিয়ে?

অনেক কিছু বানানো সম্ভব। উদাহরণস্বরূপ:

- **একটি চ্যাটবট**। আপনার কোম্পানি এবং তার পণ্য সম্পর্কিত প্রশ্নের উত্তর দেয় এমন একটি চ্যাটবট একটি ভাল উদাহরণ হতে পারে।
- **সহকারী**। LLM গুলো যেমন টেক্সট সারসংক্ষেপ করা, টেক্সট থেকে তথ্য আহরণ করা, রিজিউমে তৈরি করা ইত্যাদি কাজে দারুণ।
- **কোড সহকারী**। আপনি যে ভাষা মডেল ব্যবহার করেন তার উপর ভিত্তি করে, আপনি এমন একটি কোড সহকারী তৈরি করতে পারেন যা কোড লিখতে সাহায্য করে। উদাহরণস্বরূপ, আপনি GitHub Copilot বা ChatGPT ব্যবহার করে কোড লেখার সাহায্য পেতে পারেন।

## কিভাবে শুরু করব?

ঠিক আছে, আপনাকে একটি LLM এর সাথে ইন্টিগ্রেট করার উপায় খুঁজে বের করতে হবে যা সাধারণত নিম্নলিখিত দুইটি পদ্ধতি অনুসরণ করে:

- একটি API ব্যবহার করুন। এখানে আপনি আপনার প্রম্পট সহ ওয়েব রিকোয়েস্ট তৈরি করবেন এবং তৈরি হওয়া টেক্সট পাবেন।
- একটি লাইব্রেরি ব্যবহার করুন। লাইব্রেরিগুলি API কলগুলি enkapsulate করে এবং ব্যবহারে সহজ করে।

## লাইব্রেরি/SDK

LLM এর সাথে কাজ করার জন্য কিছু পরিচিত লাইব্রেরি আছে যেমন:

- **openai**, এই লাইব্রেরি সহজভাবে আপনার মডেলের সাথে সংযোগ স্থাপন এবং প্রম্পট পাঠাতে সাহায্য করে।

এরপর কিছু উচ্চতর স্তরের লাইব্রেরি আছে যেমন:

- **Langchain**। Langchain পরিচিত এবং Python সমর্থন করে।
- **Semantic Kernel**। Semantic Kernel হলো Microsoft এর একটি লাইব্রেরি যা C#, Python, এবং Java ভাষা সমর্থন করে।

## openai ব্যবহার করে প্রথম অ্যাপ

চলুন দেখি কিভাবে আমাদের প্রথম অ্যাপ তৈরি করতে হয়, কোন লাইব্রেরি দরকার, কতটা প্রয়োজন ইত্যাদি।

### openai ইনস্টল করুন

  > [!NOTE] Codespaces বা Devcontainer এ যদি এই নোটবুকটি চালান তবে এই ধাপটি প্রয়োজন নেই


OpenAI বা Azure OpenAI এর সাথে ইন্টারঅ্যাক্ট করার জন্য অনেক লাইব্রেরি আছে। C#, Python, JavaScript, Java এবং আরও অনেক প্রোগ্রামিং ভাষা ব্যবহার করা সম্ভব।
আমরা `openai` Python লাইব্রেরি ব্যবহার করেছি, তাই আমরা `pip` ব্যবহার করে এটি ইনস্টল করব।

```bash
pip install openai
```

আপনি যদি Codespaces বা Dev Container এ এই নোটবুক চালাচ্ছেন না, তবে আপনাকে আপনার মেশিনে [Python](https://www.python.org/) ইনস্টল করতেও হবে।

### একটি রিসোর্স তৈরি করুন এবং আপনার API কী সন্ধান করুন

যদি আপনি এখনও না করে থাকেন, তাহলে নিম্নলিখিত ধাপগুলি সম্পাদন করতে হবে:

- OpenAI এ একটি অ্যাকাউন্ট তৈরি করুন <https://platform.openai.com/signup>।
- এখন আপনার API কী সংগ্রহ করুন <https://platform.openai.com/api-keys>।

>[!NOTE]
> আপনার API কী আলাদা রাখা বাঞ্ছনীয়। আপনি পরিবেশ ভেরিয়েবল ব্যবহার করে এটি করতে পারেন।
> - আপনার `.env` ফাইলে `OPENAI_KEY` পরিবেশ ভেরিয়েবলটি আপনার API কী দিয়ে সেট করুন। আপনি যদি এই কোর্সের পূর্ববর্তী অনুশীলনগুলি সম্পন্ন করে থাকেন, তাহলে আপনি প্রস্তুত।
> - গুরুত্বপূর্ণ যে API কী একবারই অ্যাক্সেসযোগ্য হবে। তাই এটি সঠিকভাবে কপি করা হয়েছে কিনা যাচাই করা গুরুত্বপূর্ণ। যদি এটি কাজ না করে, তাহলে কী মুছে ফেলুন এবং নতুন একটি তৈরি করুন।


### OpenAI কনফিগারেশন সেটআপ করুন

আপনি যদি OpenAI ব্যবহার করেন, তাহলে কনফিগারেশন সেটআপ কেমন হয় তা এখানে দেখুন:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

উপরের কোডে আমরা নিম্নলিখিত সেট করছি:

- `api_key`, এটি আপনার API কী যা OpenAI ড্যাশবোর্ডে পাওয়া যাবে।
- `deployment`, এটি আপনার GPT সংস্করণ।

> [!NOTE]
> `os.environ` একটি ফাংশন যা পরিবেশ ভেরিয়েবল পড়ে। আপনি এর মাধ্যমে `OPENAI_API_KEY` এর মত পরিবেশ ভেরিয়েবল পড়তে পারেন।

## টেক্সট তৈরি করুন

টেক্সট তৈরি করার উপায় হল `responses.create` মেথড ব্যবহার করা। একটি উদাহরণ দেখুন:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

উপরের কোডে, আমরা একটি রেসপন্স অবজেক্ট তৈরি করেছি এবং আমাদের মডেল এবং প্রম্পট পাস করেছি। তারপর আমরা তৈরি হওয়া টেক্সট প্রিন্ট করেছি।

### চ্যাট রেসপন্স

এপর্যন্ত, আপনি দেখেছেন কীভাবে একক প্রম্পট থেকে টেক্সট তৈরি করা যায়। Responses API চ্যাটবটের জন্যও ভাল-পর্যাপ্ত— আপনি একটি মেসেজের তালিকা `input` হিসেবে প্রদান করেন যা সংলাপ গঠনে সাহায্য করে। এটি ব্যবহারের একটি উদাহরণ নিচে দেওয়া হলো:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

এই ফাংশনালিটির সম্পর্কে আরও তথ্য একটি আগাম অধ্যায়ে দেওয়া হবে।

1. একটি ভার্চুয়াল পরিবেশ তৈরি করুন এবং openai ইনস্টল করুন:

  > [!NOTE] যদি আপনি এই নোটবুকটি Codespaces বা Devcontainer-এর মধ্যে চালান তাহলে এই ধাপটি প্রয়োজনীয় নয়

## বিভিন্ন ধরণের প্রম্পট, বিভিন্ন কাজের জন্য

এখন আপনি দেখেছেন কিভাবে একটি প্রম্পট ব্যবহার করে টেক্সট তৈরি করা যায়। আপনার কাছে এমন একটি প্রোগ্রামও চালু রয়েছে যা আপনি পরিবর্তন করে বিভিন্ন ধরণের টেক্সট তৈরি করতে পারেন।

প্রম্পট বিভিন্ন কাজের জন্য ব্যবহার করা যেতে পারে। উদাহরণস্বরূপ:

- **এক ধরনের টেক্সট তৈরি করা**। যেমন, আপনি একটি কবিতা, কুইজের জন্য প্রশ্ন ইত্যাদি তৈরি করতে পারেন।
- **তথ্য অনুসন্ধান করা**। আপনি প্রম্পট ব্যবহার করে তথ্য খুঁজে পেতে পারেন যেমন নিম্নলিখিত উদাহরণ 'ওয়েব ডেভেলপমেন্টে CORS মানে কী?'।
- **কোড তৈরি করা**। আপনি প্রম্পট ব্যবহার করে কোড তৈরি করতে পারেন, যেমন ইমেল যাচাইয়ের জন্য একটি নিয়মিত অভিব্যক্তি (regex) তৈরি করা বা কেন নয়, একটি সম্পূর্ণ প্রোগ্রাম, যেমন একটি ওয়েব অ্যাপ তৈরি করা?

## ব্যায়াম - একটি রেসিপি জেনারেটর তৈরি করুন

এখন যেহেতু আমরা একটি দৃশ্যপট অনুশীলন করেছি, চলুন প্রদর্শিত দৃশ্যপটের সাথে মেলে এমন কোড লিখি। এর জন্য, নিম্নলিখিত ধাপগুলি অনুসরণ করুন:

১. বিদ্যমান *app.py* ফাইলটি একটি শুরু পয়েন্ট হিসাবে ব্যবহার করুন
২. `prompt` পরিবর্তনশীলটি খুঁজে বের করুন এবং এর কোডটি নিম্নরূপ পরিবর্তন করুন:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

দারুণ, চলুন দেখি কিভাবে আমরা বিষয়গুলো উন্নত করতে পারি। বিষয়গুলো উন্নত করার জন্য, আমরা চাই কোডটি নমনীয় হোক, যাতে উপকরণ এবং রেসিপির সংখ্যা উন্নত করা এবং পরিবর্তন করা যায়।


১. চলুন কোডটি নিম্নরূপ পরিবর্তন করি:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### ফিল্টার এবং শপিং লিস্ট যোগ করে উন্নত করুন

এখন আমাদের কাছে এমন একটি কাজের অ্যাপ আছে যা রেসিপি তৈরি করতে সক্ষম এবং এটি নমনীয় কারণ এটি ব্যবহারকারীর ইনপুটের উপর নির্ভর করে, রেসিপির সংখ্যার পাশাপাশি ব্যবহৃত উপাদানগুলোর উপরও।

এটিকে আরও উন্নত করতে, আমরা নিম্নলিখিত যোগ করতে চাই:

- **উপাদান ফিল্টার করা**। আমরা এমন উপাদানগুলি ফিল্টার করতে চাই যা আমরা পছন্দ করি না বা যেগুলোতে আমরা এলার্জি। এই পরিবর্তনটি সম্পন্ন করার জন্য, আমরা আমাদের বিদ্যমান প্রম্পট সম্পাদনা করতে পারি এবং এর শেষে একটি ফিল্টার শর্ত যোগ করতে পারি এরকম:


    উপরোক্ত কোডে, আমরা প্রম্পটের শেষে `{filter}` যোগ করেছি এবং ব্যবহারকারীর কাছ থেকে ফিল্টার মান সংগ্রহ করেছি।

    যেমন আপনি দেখতে পাচ্ছেন, দুধ (milk) যুক্ত রেসিপিগুলো ফিল্টার আউট হয়েছে। তবে, যদি আপনি ল্যাকটোজ অসহিষ্ণু হন, তাহলে আপনি পনির (cheese) যুক্ত রেসিপিগুলোও ফিল্টার করতে চাইতে পারেন, তাই এটা পরিষ্কার করা দরকার।

- **একটি শপিং লিস্ট তৈরি করুন**। আমরা একটি শপিং লিস্ট উৎপাদন করতে চাই, যা বিবেচনা করবে যে আমাদের বাড়িতে কি আছে।

    এই ফাংশনালিটির জন্য, আমরা হয় সবকিছু একসাথে একটি প্রম্পটে সমাধান করার চেষ্টা করতে পারি বা দুইটি প্রম্পটে ভাগ করতে পারি। চলুন দ্বিতীয় পদ্ধতি ট্রাই করি। এখানে আমরা অতিরিক্ত একটি প্রম্পট যোগ করার পরামর্শ দিচ্ছি, তবে সেটি কাজ করার জন্য, আমাদের প্রথম প্রম্পটের ফলাফলের উপর ভিত্তি করে দ্বিতীয় প্রম্পটে প্রসঙ্গ যোগ করতে হবে।

    প্রথম প্রম্পটের ফলাফল আউটপুট করার অংশ কোডের মধ্যে খুঁজে বের করুন এবং নিচের কোডটি নিচে যোগ করুন:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # উত্তর মুদ্রণ করুন
    print("Shopping list:")
    print(response.output_text)
    ```

    নিম্নলিখিতগুলি লক্ষ করুন:

    - আমরা একটি নতুন প্রম্পট গঠন করছি, যেখানে প্রথম প্রম্পটের ফলাফল নতুন প্রম্পটে যোগ করা হচ্ছে:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - আমরা একটি নতুন অনুরোধ পাঠাচ্ছি, তবে প্রথম প্রম্পটে আমরা কতগুলো টোকেন চাইছিলাম তা বিবেচনা করেও, তাই এবার আমরা `max_output_tokens` 1200 দিয়েছি। 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        এই কোডটি চালানো হলে নিম্নলিখিত আউটপুট পাওয়া যায়:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.
        
        -Apple and flour pancakes: 1 cup flour, 1/2 tsp baking powder, 1/2 tsp baking soda, 1/4 tsp salt, 1 tbsp sugar, 1 egg, 1 cup buttermilk or sour milk, 1/4 cup melted butter, 1 Granny Smith apple, peeled and grated
        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **টোকেন দৈর্ঘ্য সম্পর্কে একটি কথা**। আমাদের তৈরি করতে চাওয়া টেক্সটের জন্য কতগুলো টোকেন লাগবে তা বিবেচনা করা উচিত। টোকেনের জন্য টাকা খরচ হয়, তাই যতটা সম্ভব আমরা ব্যবহৃত টোকেনের পরিমাণ সাশ্রয়ী করার চেষ্টা করা উচিত। উদাহরণস্বরূপ, প্রম্পট এমনভাবে সাজাতে পারি যাতে কম টোকেন ব্যবহার হয়?

   টোকেন পরিবর্তন করতে, আপনি `max_output_tokens` প্যারামিটার ব্যবহার করতে পারেন। উদাহরণস্বরূপ, যদি আপনি ১০০ টোকেন ব্যবহার করতে চান, তাহলে আপনি করবেন:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **তাপমাত্রা নিয়ে পরীক্ষা-নিরীক্ষা**। তাপমাত্রা এমন একটি বিষয় যা এখন পর্যন্ত আমরা উল্লেখ করি নি, কিন্তু এটি আমাদের প্রোগ্রামের পারফরম্যান্সের জন্য একটি গুরুত্বপূর্ণ প্রসঙ্গ। তাপমাত্রার মান যত বেশি হবে আউটপুট তত বেশি র্যান্ডম হবে। বিপরীতে, তাপমাত্রার মান যত কম হবে আউটপুট তত বেশি পূর্বনির্ধারিত বা প্রত্যাশিত হবে। বিবেচনা করুন আপনি কি আউটপুটে বৈচিত্র্য চান কি না।

   তাপমাত্রা পরিবর্তন করতে, আপনি `temperature` প্যারামিটার ব্যবহার করতে পারেন। উদাহরণস্বরূপ, আপনি 0.5 তাপমাত্রা ব্যবহার করতে চাইলে:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > লক্ষণীয়, 1.0 এর কাছাকাছি মানে আউটপুট বেশি বৈচিত্র্যময় হবে।



## অ্যাসাইনমেন্ট

এই অ্যাসাইনমেন্টের জন্য, আপনি যা তৈরি করবেন তা বেছে নিতে পারেন।

এখানে কিছু পরামর্শ:

- রেসিপি জেনারেটর অ্যাপটিকে আরও উন্নত করুন। তাপমাত্রার মান এবং প্রম্পটগুলো নিয়ে খেলে দেখুন আপনি কী তৈরি করতে পারেন।
- "স্টাডি বাডি" তৈরি করুন। এই অ্যাপটি একটি বিষয় সম্পর্কে প্রশ্নের উত্তর দিতে সক্ষম হওয়া উচিত, উদাহরণস্বরূপ পাইথন; আপনি এমন প্রম্পট রাখতে পারেন "পাইথনে একটি নির্দিষ্ট বিষয় কী?", অথবা একটি প্রম্পট হতে পারে, আমাকে একটি নির্দিষ্ট বিষয়ের কোড দেখাও ইত্যাদি।
- ইতিহাস বট, ইতিহাসকে জীবন্ত করে তুলুন, বটকে নির্দেশ দিন একটি নির্দিষ্ট ঐতিহাসিক চরিত্রের মতো অভিনয় করতে এবং তার জীবন ও সময় সম্পর্কে প্রশ্ন করুন।
