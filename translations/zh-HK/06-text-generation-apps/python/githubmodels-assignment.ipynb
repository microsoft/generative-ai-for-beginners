## 什麼是文字生成應用程式？

通常當你建立應用程式時，它會有某種介面，例如以下：

- 基於指令的。控制台應用程式是典型的應用程式，你在其中輸入指令，它執行任務。例如，`git` 是一個基於指令的應用程式。
- 使用者介面（UI）。某些應用程式有圖形使用者介面（GUI），你可以點擊按鈕、輸入文字、選擇選項等。

### 控制台和 UI 應用程式的限制

與基於指令的應用程式相比，你需要輸入一個指令：

- <strong>它有限制</strong>。你不能隨便輸入任何指令，只有應用程式支援的指令才能使用。
- <strong>語言專用</strong>。某些應用程式支援多種語言，但預設是為特定語言建立，即使你可以添加更多語言支援。

### 文字生成應用程式的優點

那麼文字生成應用程式有何不同？

在文字生成應用程式中，你具有更多靈活性，你不會被限於一組指令或特定的輸入語言。相反地，你可以使用自然語言來與應用程式互動。另外一個好處是，因為你已經在與一個基於大量語料庫訓練的資料來源互動，而傳統應用程式可能只受限於資料庫中的內容。

### 我可以用文字生成應用程式建立什麼？

你可以建立很多東西。例如：

- <strong>聊天機器人</strong>。例如，一個能回答關於公司及產品問題的聊天機器人會是一個不錯的選擇。
- <strong>助手工具</strong>。大型語言模型擅長像是文本摘要、文本洞察分析、生成履歷等。
- <strong>程式碼助手</strong>。根據你使用的語言模型，你可以建立協助撰寫程式碼的助手。例如，你可以使用 GitHub Copilot 或 ChatGPT 來幫助撰寫程式碼。

## 練習 - 建立食譜產生器

現在我們已經演練了一個情境，讓我們撰寫程式碼來符合示範的情境。請依照以下步驟進行：

1. 使用現有的檔案作為起點
1. 建立一個 `prompt` 變數，並將範例程式碼更改如下：


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
