## 什麼是文本生成應用程式？

通常你建立應用程式會有某種界面，如下：

- 指令式。控制台應用程式是典型的應用程式，你輸入指令，它執行任務。例如 `git` 是一個指令式應用程式。
- 使用者介面（UI）。有些應用程式有圖形使用者介面（GUI），你會點擊按鈕、輸入文字、選擇選項等。

### 控制台和使用者介面應用程式的限制

比較指令式應用程式，你只能輸入指令：

- <strong>有限制</strong>。你不能輸入任意指令，只能輸入應用程式支援的指令。
- <strong>語言特定</strong>。一些應用支援多種語言，但預設是建構用於特定語言，即使你可以額外增加語言支援。

### 文本生成應用程式的好處

那文本生成應用程式有什麼不同？

在文本生成應用程式中，你有更大的彈性，不限於一組指令或特定輸入語言。你可以用自然語言與應用程式互動。另一個好處是，由於你已經與一個訓練於大量文本語料庫的資料來源互動，傳統應用程式可能受限於資料庫中的資料，但文本生成應用程式則可突破此限制。

### 我可以用文本生成應用程式建立什麼？

你可以建立許多類型的應用，例如：

- <strong>聊天機器人</strong>。回答有關主題的問題，如你的公司及其產品，是不錯的選擇。
- <strong>助手</strong>。大型語言模型（LLMs）擅長摘要文本、從文本獲得見解、產生如履歷表等文本內容。
- <strong>程式碼助手</strong>。依據你使用的語言模型，你可以建立幫助編寫程式碼的助手。例如，可以使用 GitHub Copilot 或 ChatGPT 協助撰寫程式碼。

## 文本生成

產生文本的方法是使用 `responses.create` 函數。範例如下：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上面程式碼中，我們建立一個回應物件，並傳入想使用的模型和提示，接著印出生成的文字。

### 聊天回應

## 練習 - 你的第一個文本生成應用程式

現在我們已學會如何設定及配置 OpenAI 服務，是時候建立你的第一個文本生成應用程式。請依照以下步驟完成：


1. 建立虛擬環境並安裝 openai：

  > [!NOTE] 如果你在 Codespaces 或 Devcontainer 內運行此筆記本，這步驟並非必要


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> 如果你使用 Windows，請輸入 `venv\Scripts\activate`，而不是 `source venv/bin/activate`。

> [!NOTE]
> 前往 https://platform.openai.com/settings/organization/api-keys 找到你的 OpenAI 金鑰，並搜尋 `API keys`。你可以在那裡創建新的金鑰並即時複製該值。


1. 建立一個 *app.py* 檔案並且輸入以下程式碼：


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    你應該會看到類似以下的輸出：

    ```output

    ```
