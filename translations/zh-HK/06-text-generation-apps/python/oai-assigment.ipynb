## 什麼是文字生成應用程式？

通常當你建立一個應用程式，它會有某種介面，例如以下幾類：

- 命令式（Command-based）。控制台應用程式是典型類型，你輸入指令，它會執行任務。例如，`git` 就是一個命令式應用程式。
- 使用者介面（UI）。有些應用程式具備圖形使用者介面（GUI），讓你可點擊按鈕、輸入文字、選擇選項等等。

### 控制台和使用者介面應用程式的限制

與命令式應用程式相比，你會輸入指令：

- <strong>有其限制</strong>。你不能隨意輸入任何指令，只能使用該應用程式支援的指令。
- <strong>語言特定</strong>。有些應用程式支援多種語言，但預設是為特定語言打造，儘管你可以額外增加語言支援。

### 文字生成應用程式的好處

那麼文字生成應用程式又有什麼不同呢？

在文字生成應用程式中，你會有更多彈性，不受限於一組指令集或特定輸入語言。你可以使用自然語言與應用程式互動。另一個好處是，你正與經過龐大語料庫訓練的資料源互動，而傳統應用程式可能只限於資料庫中的內容。

### 我可以用文字生成應用程式建立什麼？

你可以建立很多東西，例如：

- <strong>聊天機器人</strong>。一個針對主題問題（例如你的公司及產品）作答的聊天機器人可能非常合適。
- <strong>助理</strong>。大型語言模型（LLMs）擅長總結文字、從文字中取得洞見、產生如履歷等文本。
- <strong>程式碼助理</strong>。依你使用的語言模型，可以建立協助寫程式碼的助理。例如，你可以使用 GitHub Copilot 以及 ChatGPT 來幫助寫程式碼。

## 如何開始？

你需要找到與大型語言模型（LLM）整合的方法，通常有以下兩種做法：

- 使用 API。這裡你會構造帶有提示語的網路請求，並獲取生成的文字回應。
- 使用函式庫。函式庫協助封裝 API 調用，使用起來更簡易。

## 生成文字

生成文字的方式是使用 `responses.create` 方法。這裡有一個範例：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上述程式碼中，我們建立了一個 response 物件，並傳入要使用的模型及提示語。然後我們列印出生成的文字。

### 聊天回應

到目前為止，你已見過如何針對單一提示產生文字。Responses API 也非常適合用於聊天機器人——你可以提供一串訊息作為 `input`，用以建立對話。下面是使用範例：


## 練習 - 你的第一個文字生成應用程式

現在我們已學會如何設定及配置 OpenAI 服務，是時候來建立你的第一個文字生成應用程式了。要建置你的應用程式，請按以下步驟操作：


1. 建立虛擬環境並安裝 openai：

  > [!NOTE] 如果你在 Codespaces 或 Devcontainer 中運行此筆記本，此步驟不是必須的


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]

> [!NOTE]
> 你可以到 https://platform.openai.com/settings/organization/api-keys  找到你的 OpenAI 金鑰，並搜尋 `API keys`。你可以在那裏創建一個新的金鑰並立即複製其值。


1. 建立一個 *app.py* 檔案，並加入以下程式碼：


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

    ```output

    ```
