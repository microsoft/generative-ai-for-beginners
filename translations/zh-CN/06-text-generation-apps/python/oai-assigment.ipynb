## 什么是文本生成应用？

通常，当你构建一个应用时，会有某种界面，例如：

- 基于命令的。控制台应用是典型的基于命令的应用，你输入一个命令，程序执行任务。例如，`git` 是基于命令的应用。
- 用户界面（UI）。一些应用有图形用户界面（GUI），你可以点击按钮、输入文本、选择选项等。

### 控制台和 UI 应用的局限性

比较一下基于命令的应用，你需要输入一个命令：

- <strong>有限制</strong>。你不能输入任意命令，只有应用支持的命令可用。
- <strong>特定语言</strong>。有些应用支持多语言，但默认情况下应用是针对特定语言构建的，尽管你也可以添加更多语言支持。

### 文本生成应用的优点

那么文本生成应用有什么不同呢？

在文本生成应用中，你有更大的灵活性，不受限于固定的命令集或特定输入语言。相反，你可以使用自然语言与应用交互。另一个优点是你实际上是在与一个已经经过庞大语料库训练的数据源交互，而传统应用的数据通常仅限于数据库里的内容。

### 我可以用文本生成应用构建什么？

你可以构建很多东西。例如：

- <strong>聊天机器人</strong>。一个回答关于主题问题的聊天机器人，比如你的公司和产品，非常合适。
- <strong>助手</strong>。大型语言模型很擅长总结文本、获取文本见解、生成文本（如简历）等。
- <strong>代码助手</strong>。根据你使用的语言模型，你可以构建帮助写代码的代码助手。例如，你可以使用 GitHub Copilot 和 ChatGPT 来帮你写代码。

## 生成文本

生成文本的方法是使用 `responses.create` 方法。示例如下：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

上述代码中，我们创建了一个响应对象，并传入想使用的模型和提示词，然后打印生成的文本。

### 聊天响应

到目前为止，你已经看到如何从单个提示词生成文本。Responses API 也非常适合聊天机器人——你提供一组消息作为 `input` 来构建对话。示例如下：


1. 创建一个虚拟环境并安装 openai：

  > [!NOTE] 如果您在 Codespaces 上或 Devcontainer 中运行此笔记本，则此步骤不是必需的

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````
