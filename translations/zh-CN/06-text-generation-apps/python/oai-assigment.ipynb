## 什么是文本生成应用？

通常，当你构建一个应用时，它会有某种界面，比如：

- 基于命令的。控制台应用是典型的此类应用，你输入一个命令，应用执行任务。例如，`git` 就是一个基于命令的应用。
- 用户界面（UI）。一些应用有图形用户界面（GUI），你可以点击按钮、输入文本、选择选项等。

### 控制台和 UI 应用的限制

与此相比，基于命令的应用中你输入命令：

- <strong>它是有限的</strong>。你不能输入任何命令，只能输入应用支持的命令。
- <strong>语言特定</strong>。有些应用支持多种语言，但默认构建针对特定语言，即使可以添加更多语言支持。

### 文本生成应用的优势

那文本生成应用有什么不同呢？

文本生成应用更灵活，不限于一组命令或特定输入语言。相反，你可以使用自然语言与应用交互。另一个好处是，你已经在与经过大量信息语料库训练的数据源交互，相较之下传统应用可能只受限于数据库中的内容。

### 我可以用文本生成应用构建什么？

你可以构建许多东西。例如：

- <strong>聊天机器人</strong>。一个回答关于话题、公司及其产品问题的聊天机器人可能非常合适。
- <strong>助手</strong>。大型语言模型在文本摘要、获取文本洞察、生成简历等方面非常出色。
- <strong>代码助手</strong>。根据所用语言模型，你可以构建一个帮助你编写代码的代码助手。例如，你可以使用 GitHub Copilot 或 ChatGPT 来帮助编写代码。

## 生成文本

生成文本的方式是使用 `responses.create` 方法。示例如下：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上面的代码中，我们创建了一个响应对象，传入我们想用的模型和提示。然后打印生成的文本。

### 聊天响应

    找到代码中打印第一个提示结果的部分，并在其下添加以下代码：
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # 打印响应
    print("Shopping list:")
    print(response.output_text)
    ```

    注意以下几点：

    - 我们通过将第一个提示的结果添加到新提示中来构造新提示：
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - 我们进行新的请求，同时考虑第一次提示中请求的令牌数，所以这次 `max_output_tokens` 设置为1200。

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```

- <strong>关于令牌长度</strong>。我们应考虑需要多少令牌来生成所需文本。令牌会产生费用，因此应尽量节约令牌使用量。例如，我们是否可以改写提示，使其使用更少的令牌？

   要更改使用的令牌数，可以使用 `max_output_tokens` 参数。例如，如果想使用100个令牌，可以这么做：

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- <strong>尝试调整温度</strong>。温度是我们尚未提及但对程序性能非常重要的参数。温度值越高，输出越随机；温度值越低，输出越可预测。考虑你是否希望输出有变化。

   要调整温度，可以使用 `temperature` 参数。例如，如果想使用0.5的温度，可以这么做：

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```
