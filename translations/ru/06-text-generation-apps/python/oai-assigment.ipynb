## Генерация текста

Для генерации текста используется метод `responses.create`. Вот пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В этом коде мы создаём объект ответа, передаём модель и промпт, затем выводим сгенерированный текст.

### Ответы в чате

До сих пор вы видели, как можно генерировать текст из одного промпта. API Responses также хорошо подходит для чатботов — вы передаёте список сообщений в `input`, чтобы построить диалог. Вот пример использования:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Теперь, когда мы проиграли сценарий, давайте напишем код, соответствующий показанному сценарию. Для этого выполните следующие шаги:

1. Используйте существующий файл *app.py* в качестве отправной точки
1. Найдите переменную `prompt` и измените её код на следующий:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Отлично, давайте посмотрим, как мы можем улучшить ситуацию. Чтобы сделать код лучше, мы хотим убедиться, что он гибкий, чтобы ингредиенты и количество рецептов можно было улучшать и изменять.


1. Давайте изменим код следующим образом:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Улучшаем, добавив фильтр и список покупок

Чтобы улучшить его, мы хотим добавить следующее:

- **Фильтрация ингредиентов**. Мы хотим иметь возможность фильтровать ингредиенты, которые нам не нравятся или вызывают аллергию. Чтобы выполнить это изменение, мы можем отредактировать существующий запрос и добавить к нему в конце условие фильтрации, как показано ниже:

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
