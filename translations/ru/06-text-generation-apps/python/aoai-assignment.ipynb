## Что такое приложение для генерации текста?

Обычно при создании приложения оно имеет какой-то интерфейс, например:

- Основанный на командах. Консольные приложения — это типичные приложения, где вы вводите команду, и она выполняет задачу. Например, `git` — это приложение основанное на командах.
- Пользовательский интерфейс (UI). Некоторые приложения имеют графические интерфейсы (GUI), где можно нажимать кнопки, вводить текст, выбирать опции и многое другое.

### Консольные и UI приложения имеют ограничения

Сравните это с приложением на основе команд, где вы вводите команду:

- **Ограничено**. Вы не можете вводить любую команду, только те, которые поддерживает приложение.
- **Языково специфично**. Некоторые приложения поддерживают несколько языков, но по умолчанию приложение создаётся под конкретный язык, даже если можно добавить поддержку других.

### Преимущества приложений генерации текста

Чем же приложение генерации текста отличается?

В приложении генерации текста у вас больше гибкости — вы не ограничены набором команд или конкретным языком ввода. Вместо этого вы можете использовать естественный язык для взаимодействия с приложением. Ещё одно преимущество в том, что вы уже взаимодействуете с источником данных, который обучен на огромном количестве информации, в то время как традиционное приложение может быть ограничено тем, что находится в базе данных.

### Настройка конфигурации Azure

Если вы используете Azure OpenAI, вот как настроить конфигурацию. API откликов обслуживается с конечной точки Azure OpenAI **v1**, поэтому указываем клиенту `OpenAI` конечную точку `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

В приведённом выше примере мы настраиваем следующее:

- `api_key` — это ваш ключ API, найденный в портале Azure.
- `base_url` — это конечная точка Azure OpenAI с добавлением `/openai/v1/`. Вы можете найти конечную точку в портале Azure рядом с вашим ключом API. Использование конечной точки v1 означает, что передавать `api_version` больше не требуется.
- `deployment` — это имя развертывания модели, которое вы создали в портале Foundry.

> [!NOTE]
> `os.environ` — это отображение, которое считывает переменные окружения. Вы можете использовать его для чтения переменных, таких как `AZURE_OPENAI_API_KEY` и `AZURE_OPENAI_ENDPOINT`.

## Генерация текста

Для генерации текста используется метод `responses.create`. Вот пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В приведённом коде мы создаём объект отклика и передаём модель, которую хотим использовать, а также подсказку. Затем выводим сгенерированный текст.

### Ответы чат-бота

## Упражнение - создать генератор рецептов

Теперь, когда мы разыграли сценарий, давайте напишем код, соответствующий продемонстрированному сценарию. Для этого выполните следующие шаги:

1. Используйте существующий файл *app.py* в качестве отправной точки
1. Найдите переменную `prompt` и измените её код следующим образом:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Отлично, давайте посмотрим, как мы можем улучшить ситуацию. Чтобы улучшить всё, нам нужно сделать код гибким, чтобы ингредиенты и количество рецептов можно было улучшать и изменять.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list
