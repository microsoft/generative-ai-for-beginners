## Что такое приложение для генерации текста?

Обычно, когда вы создаёте приложение, у него есть какой-то интерфейс, например:

- Командное. Консольные приложения — это типичные приложения, где вы вводите команду, и она выполняет задачу. Например, `git` — это приложение с командным интерфейсом.
- Пользовательский интерфейс (UI). Некоторые приложения имеют графический интерфейс, где вы кликаете кнопки, вводите текст, выбираете опции и так далее.

### Консольные и UI-приложения имеют ограничения

Сравним с командным приложением, где вы вводите команду:

- **Ограниченное**. Вы не можете просто ввести любую команду, только те, которые поддерживает приложение.
- **Язык специфический**. Некоторые приложения поддерживают много языков, но по умолчанию приложение создано для конкретного языка, даже если вы можете добавить поддержку других.

### Преимущества приложений для генерации текста

Так в чём же разница у приложения для генерации текста?

В приложении для генерации текста у вас больше гибкости: вы не ограничены набором команд или конкретным языком ввода. Вместо этого вы можете использовать естественный язык для взаимодействия с приложением. Ещё одно преимущество в том, что при взаимодействии с источником данных, который обучен на огромном корпусе информации, возможности шире по сравнению с традиционным приложением, ограниченным содержимым базы данных.

### Что можно создать с помощью приложения для генерации текста?

### Настройка конфигурации Azure

Если вы используете Azure OpenAI, вот как настроить конфигурацию. API ответов обслуживается с Azure OpenAI **endpoints версии v1**, поэтому мы указываем клиенту `OpenAI` `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Выше мы устанавливаем следующее:

- `api_key` — это ваш ключ API, найденный в портале Azure.
- `base_url` — это ваш endpoint Azure OpenAI с добавленным `/openai/v1/`. Endpoint можно найти в портале Azure рядом с вашим ключом API. Использование v1 endpoint означает, что больше не нужно передавать `api_version`.
- `deployment` — это имя развернутой модели, которую вы создали в портале Foundry.

> [!NOTE]
> `os.environ` — это отображение, которое читает переменные окружения. С его помощью можно читать переменные окружения, такие как `AZURE_OPENAI_API_KEY` и `AZURE_OPENAI_ENDPOINT`.

## Генерация текста

Для генерации текста используйте метод `responses.create`. Вот пример:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

В приведённом выше коде мы создаём объект ответа, передаём модель и запрос, а затем выводим сгенерированный текст.

## Упражнение — создать генератор рецептов

Теперь, когда мы проиграли сценарий, давайте напишем код, соответствующий продемонстрированному сценарию. Для этого выполните следующие шаги:

1. Используйте существующий файл *app.py* в качестве отправной точки
1. Найдите переменную `prompt` и измените ее код на следующий:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Отлично, давайте посмотрим, как можно улучшить ситуацию. Чтобы улучшить всё, мы хотим убедиться, что код гибкий, чтобы ингредиенты и количество рецептов можно было улучшать и изменять.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list
