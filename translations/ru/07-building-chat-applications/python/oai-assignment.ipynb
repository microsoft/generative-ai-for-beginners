### 2. Импорт вспомогательных библиотек и создание учетных данных


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Поиск подходящей модели  
Модели, такие как GPT-4o и GPT-4o mini, могут понимать и генерировать естественный язык и доступны на платформе OpenAI с разными уровнями мощности и скорости, подходящими для различных задач.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Дизайн подсказок  

«Магия больших языковых моделей заключается в том, что, обучаясь минимизировать ошибку предсказания на огромных объемах текста, модели в итоге изучают концепции, полезные для этих предсказаний. Например, они учатся концепциям, таким как»(1):

* как правильно писать
* как работает грамматика
* как перефразировать
* как отвечать на вопросы
* как вести разговор
* как писать на многих языках
* как программировать
* и так далее.

#### Как управлять большой языковой моделью  
«Из всех входных данных для большой языковой модели самым влиятельным по праву считается текстовая подсказка»(1).

Большие языковые модели можно заставить генерировать результат несколькими способами:

Инструкция: Скажите модели, чего вы хотите
Завершение: Побудите модель завершить начало того, что вы хотите
Демонстрация: Покажите модели, чего вы хотите, с помощью:
Нескольких примеров в подсказке
Сотен или тысяч примеров в наборе данных для дообучения»



#### Есть три основных правила создания подсказок:

**Показывайте и объясняйте**. Ясно указывайте, чего вы хотите, через инструкции, примеры или их сочетание. Если вы хотите, чтобы модель расположила список элементов в алфавитном порядке или классифицировала параграф по эмоциям, покажите ей, что именно вы ожидаете.

**Обеспечьте качество данных**. Если вы пытаетесь построить классификатор или заставить модель следовать определенному шаблону, убедитесь, что примеров достаточно. Обязательно проверяйте свои примеры — модель обычно достаточно умна, чтобы распознать базовые орфографические ошибки и выдать ответ, но также может предположить, что это сделано намеренно, что может повлиять на ответ.

**Проверьте настройки.** Параметры temperature и top_p контролируют, насколько детерминирована модель при генерации ответа. Если вы запрашиваете ответ, где есть только один правильный вариант, то их стоит установить ниже. Если вы хотите получить более разнообразные ответы, установите их выше. Главная ошибка пользователей — считать эти параметры контролем «умности» или «креативности».


Источник: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Отправить!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
