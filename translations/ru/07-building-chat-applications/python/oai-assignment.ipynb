### 2. Импортировать вспомогательные библиотеки и создать учетные данные


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Поиск подходящей модели  
Модели, такие как GPT-4o и GPT-4o mini, могут понимать и генерировать естественный язык, и доступны на платформе OpenAI с разным уровнем мощности и скорости, подходящими для разных задач.


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

## 4. Проектирование подсказок  

«Магия больших языковых моделей заключается в том, что, будучи обученными минимизировать ошибку предсказания на огромных количествах текста, модели в итоге учатся концепциям, полезным для этих предсказаний. Например, они учатся таким концепциям, как»(1):

* как правильно писать
* как работает грамматика
* как перефразировать
* как отвечать на вопросы
* как вести разговор
* как писать на многих языках
* как программировать
* и т.д.

#### Как управлять большой языковой моделью  
«Из всех входных данных для большой языковой модели самый влиятельный — это текстовая подсказка»(1).

Большие языковые модели можно направлять на создание вывода несколькими способами:

Инструкция: Скажите модели, что вы хотите
Завершение: Побудите модель завершить начало того, что вы хотите
Демонстрация: Покажите модели, что вы хотите, с помощью:
Нескольких примеров в подсказке
Сотен или тысяч примеров в обучающем наборе для тонкой настройки»



#### Существует три основных принципа создания подсказок:

**Показывайте и объясняйте**. Сделайте ясно, что вы хотите, либо через инструкции, примеры или их сочетание. Если вы хотите, чтобы модель отсортировала список элементов в алфавитном порядке или классифицировала абзац по настроению, покажите ей, что именно вы хотите.

**Обеспечьте качественные данные**. Если вы пытаетесь создать классификатор или хотите, чтобы модель следовала определенному шаблону, убедитесь, что примеров достаточно. Обязательно вычитывайте свои примеры — модель обычно достаточно умна, чтобы видеть простые орфографические ошибки и дать ответ, но может предположить, что это сделано намеренно, и это повлияет на ответ.

**Проверьте свои настройки.** Параметры temperature и top_p управляют тем, насколько детерминированной модель генерирует ответ. Если вы спрашиваете ответ, где есть только один правильный вариант, то лучше выставить эти параметры ниже. Если вы хотите получить более разнообразные ответы, то возможно стоит выставить их выше. Основная ошибка людей с этими настройками — считать, что они контролируют «умность» или «креативность».


Источник: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Отправить!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
