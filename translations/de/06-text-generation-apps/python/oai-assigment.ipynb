## Was ist eine Textgenerierungs-App?

Normalerweise hat eine App, die du baust, eine Art Benutzeroberfläche wie die folgende:

- Befehlsgesteuert. Konsolen-Apps sind typische Apps, bei denen du einen Befehl eingibst und eine Aufgabe ausgeführt wird. Zum Beispiel ist `git` eine befehlsgesteuerte App.
- Benutzeroberfläche (UI). Manche Apps haben grafische Benutzeroberflächen (GUIs), bei denen du Buttons klickst, Text eingibst, Optionen auswählst und mehr.

### Konsolen- und UI-Apps sind eingeschränkt

Vergleiche es mit einer befehlsgesteuerten App, bei der du einen Befehl eingibst:

- **Sie ist eingeschränkt**. Du kannst nicht beliebige Befehle eingeben, sondern nur die, die die App unterstützt.
- **Sprachspezifisch**. Manche Apps unterstützen viele Sprachen, aber standardmäßig ist die App für eine bestimmte Sprache gebaut, auch wenn du zusätzliche Sprachunterstützung hinzufügen kannst.

### Vorteile von Textgenerierungs-Apps

Wie unterscheidet sich also eine Textgenerierungs-App?

Bei einer Textgenerierungs-App hast du mehr Flexibilität, du bist nicht auf eine Reihe von Befehlen oder eine bestimmte Eingabesprache beschränkt. Stattdessen kannst du natürlichsprachlich mit der App interagieren. Ein weiterer Vorteil ist, dass du bereits mit einer Datenquelle interagierst, die auf einem umfangreichen Korpus von Informationen trainiert wurde, während eine herkömmliche App möglicherweise in dem, was in einer Datenbank enthalten ist, begrenzt ist.

### Was kann ich mit einer Textgenerierungs-App bauen?

Du kannst viele Dinge bauen. Beispielsweise:

- **Einen Chatbot**. Ein Chatbot, der Fragen zu Themen wie deinem Unternehmen und seinen Produkten beantwortet, könnte eine gute Wahl sein.
- **Helfer**. LLMs sind großartig bei Dingen wie Textzusammenfassung, dem Gewinnen von Einsichten aus Text, dem Erstellen von Texten wie Lebensläufen und mehr.
- **Code-Assistent**. Je nach verwendetem Sprachmodell kannst du einen Code-Assistenten bauen, der dir beim Schreiben von Code hilft. Zum Beispiel kannst du Produkte wie GitHub Copilot sowie ChatGPT einsetzen, um Code zu schreiben.

## Text generieren

Der Weg Text zu generieren ist über die Methode `responses.create`. Hier ist ein Beispiel:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Im obigen Code erstellen wir ein Antwortobjekt und übergeben das Modell, das wir verwenden wollen, sowie den Prompt. Dann drucken wir den generierten Text aus.

### Chat-Antworten

Bisher hast du gesehen, wie wir aus einem einzelnen Prompt Text generieren können. Die Responses API eignet sich auch gut für Chatbots – du gibst eine Liste von Nachrichten als `input` ein, um eine Konversation aufzubauen. Hier ein Beispiel, wie man sie verwendet:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Je höher der Temperaturwert, desto zufälliger wird die Ausgabe. Umgekehrt gilt: Je niedriger der Temperaturwert, desto vorhersehbarer die Ausgabe. Überlegen Sie, ob Sie Variation in Ihrer Ausgabe möchten oder nicht.

   Um die Temperatur anzupassen, können Sie den Parameter `temperature` verwenden. Zum Beispiel, wenn Sie eine Temperatur von 0.5 verwenden möchten, würden Sie so vorgehen:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Hinweis: Je näher am Wert 1.0, desto abwechslungsreicher die Ausgabe.
