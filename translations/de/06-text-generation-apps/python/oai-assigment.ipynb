## Was ist eine Textgenerierungs-App?

Normalerweise hat eine App eine Art Schnittstelle wie die folgende:

- Befehlsbasiert. Konsolen-Apps sind typische Apps, bei denen du einen Befehl eingibst und dieser eine Aufgabe ausführt. Beispiel: `git` ist eine befehlsbasierte App.
- Benutzeroberfläche (UI). Manche Apps haben grafische Benutzeroberflächen (GUIs), bei denen du Buttons klickst, Text eingibst, Optionen auswählst und mehr.

### Konsolen- und UI-Apps sind begrenzt

Vergleich es mit einer befehlsbasierten App, bei der du einen Befehl eintippst:

- **Sie ist begrenzt**. Du kannst nicht einfach jeden beliebigen Befehl eingeben, nur die vom System unterstützten.
- **Sprachspezifisch**. Manche Apps unterstützen viele Sprachen, aber standardmäßig sind sie für eine bestimmte Sprache gebaut, auch wenn man weitere Sprachunterstützung hinzufügen kann.

### Vorteile von Textgenerierungs-Apps

Was macht eine Textgenerierungs-App anders?

In einer Textgenerierungs-App hast du mehr Flexibilität, du bist nicht auf eine Menge Befehle oder eine bestimmte Eingabesprache beschränkt. Stattdessen kannst du natürliche Sprache verwenden, um mit der App zu interagieren. Ein weiterer Vorteil ist, dass du mit einer Datenquelle interagierst, die auf einem riesigen Korpus trainiert wurde, während eine traditionelle App oft auf eine Datenbank beschränkt ist.

### Was kann ich mit einer Textgenerierungs-App bauen?

Es gibt viele mögliche Anwendungen. Zum Beispiel:

- **Ein Chatbot**. Ein Chatbot, der Fragen zu Themen beantwortet, z. B. zu deinem Unternehmen und seinen Produkten, wäre eine gute Anwendung.
- **Helfer**. LLMs sind hervorragend darin, Texte zusammenzufassen, Einblicke aus Texten zu gewinnen, Texte zu erstellen, wie Lebensläufe und mehr.
- **Code-Assistent**. Je nach verwendetem Sprachmodell kannst du einen Code-Assistenten bauen, der dir beim Schreiben von Code hilft. Zum Beispiel kannst du Produkte wie GitHub Copilot oder ChatGPT verwenden, um Code zu schreiben.

Um Text zu generieren, verwendest du die Methode `responses.create`. Hier ein Beispiel:

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Experimentieren mit der Temperatur**. Die Temperatur haben wir bisher nicht erwähnt, sie ist aber ein wichtiger Kontext dafür, wie unser Programm funktioniert. Je höher der Temperaturwert, desto zufälliger wird die Ausgabe. Umgekehrt gilt: je niedriger der Wert, desto vorhersehbarer ist die Ausgabe. Überlegen Sie, ob Sie Variationen in der Ausgabe möchten.

   Um die Temperatur zu ändern, können Sie den Parameter `temperature` verwenden. Wenn Sie zum Beispiel eine Temperatur von 0,5 nutzen möchten, tun Sie folgendes:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Hinweis: Je näher am Wert 1.0, desto vielfältiger ist die Ausgabe.
