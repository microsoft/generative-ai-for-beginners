## Was ist eine Textgenerierungs-App?

Normalerweise hat eine App, die du baust, eine Art Benutzeroberfläche wie die folgende:

- Befehlsbasiert. Konsolen-Apps sind typische Apps, bei denen du einen Befehl eintippst und eine Aufgabe ausgeführt wird. Zum Beispiel ist `git` eine befehlsbasierte App.
- Benutzeroberfläche (UI). Manche Apps haben grafische Benutzeroberflächen (GUIs), bei denen du Buttons klickst, Text eingibst, Optionen auswählst und mehr.

### Konsolen- und UI-Apps sind eingeschränkt

Vergleiche das mit einer befehlsbasierten App, bei der du einen Befehl eintippst:

- **Sie ist eingeschränkt**. Du kannst nicht einfach irgendeinen Befehl eingeben, sondern nur die, die die App unterstützt.
- **Sprachspezifisch**. Manche Apps unterstützen viele Sprachen, aber standardmäßig ist die App für eine bestimmte Sprache gebaut, auch wenn du weitere Sprachunterstützung hinzufügen kannst.

### Vorteile von Textgenerierungs-Apps

Wie unterscheidet sich also eine Textgenerierungs-App?

In einer Textgenerierungs-App bist du flexibler, du bist nicht auf eine Menge Befehle oder eine bestimmte Eingabesprache beschränkt. Stattdessen kannst du natürliche Sprache verwenden, um mit der App zu interagieren. Ein weiterer Vorteil ist, dass du mit einer Datenquelle interagierst, die auf einem enorm großen Korpus an Informationen trainiert wurde, während eine traditionelle App möglicherweise auf das begrenzt ist, was in einer Datenbank gespeichert ist.

### Was kann ich mit einer Textgenerierungs-App bauen?

Es gibt viele Dinge, die du bauen kannst. Zum Beispiel:

- **Ein Chatbot**. Ein Chatbot, der Fragen zu Themen wie deinem Unternehmen und seinen Produkten beantwortet, könnte gut passen.
- **Helfer**. LLMs sind großartig darin, Texte zusammenzufassen, Einblicke aus Texten zu gewinnen, Texte wie Lebensläufe zu erstellen und mehr.
- **Code-Assistent**. Je nach verwendetem Sprachmodell kannst du einen Code-Assistenten bauen, der dir beim Schreiben von Code hilft. Du kannst zum Beispiel Produkte wie GitHub Copilot sowie ChatGPT nutzen, um dir beim Programmieren zu helfen.

### Azure-Konfiguration einrichten

Wenn du Azure OpenAI verwendest, so richtest du die Konfiguration ein. Die Responses-API wird vom Azure OpenAI **v1 Endpoint** bereitgestellt, also zeigen wir den `OpenAI`-Client auf `<dein-endpunkt>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Oben setzen wir Folgendes:

- `api_key`, das ist dein API-Schlüssel, den du im Azure-Portal gefunden hast.
- `base_url`, das ist dein Azure OpenAI-Endpunkt mit angehängtem `/openai/v1/`. Den Endpunkt findest du im Azure-Portal neben deinem API-Schlüssel. Die Verwendung des v1-Endpunkts bedeutet, dass du keine `api_version` mehr übergeben musst.
- `deployment`, das ist der Name der Modellbereitstellung, die du im Foundry-Portal erstellt hast.

> [!NOTE]
> `os.environ` ist eine Abbildung, die Umgebungsvariablen liest. Du kannst sie verwenden, um Umgebungsvariablen wie `AZURE_OPENAI_API_KEY` und `AZURE_OPENAI_ENDPOINT` zu lesen.

## Text generieren

Die Methode, um Text zu generieren, ist die `responses.create` Methode. Hier ist ein Beispiel:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Im obigen Code erzeugen wir ein Antwortobjekt und geben das Modell und den Prompt an. Danach geben wir den generierten Text aus.

### Chat-Antworten

Die Responses-API eignet sich sowohl für einfache Textgenerierung mit nur einem Durchlauf als auch für mehrstufige Chatbots – du fügst einfach mehr Nachrichten in die `input`-Liste ein, um ein Gespräch aufzubauen:

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  
