#### Wie man ein großes Sprachmodell steuert  
„Von allen Eingaben an ein großes Sprachmodell ist bei weitem der einflussreichste die Texteingabe(1).

Große Sprachmodelle können auf verschiedene Weise dazu gebracht werden, Ausgaben zu erzeugen:

- Anweisung: Sage dem Modell, was du möchtest
- Vervollständigung: Veranlasse das Modell, den Anfang dessen zu vervollständigen, was du möchtest
- Demonstration: Zeige dem Modell, was du möchtest, entweder mit:
- Ein paar Beispielen im Prompt
- Hunderten oder Tausenden von Beispielen in einem Fine-Tuning-Trainingsdatensatz



#### Es gibt drei grundlegende Richtlinien zur Erstellung von Prompts:

**Zeigen und sagen**. Mache klar, was du möchtest, entweder durch Anweisungen, Beispiele oder eine Kombination aus beidem. Wenn du möchtest, dass das Modell eine Liste von Elementen alphabetisch sortiert oder einen Absatz nach Stimmung klassifiziert, zeige ihm, dass das dein Wunsch ist.

**Qualitativ hochwertige Daten bereitstellen**. Wenn du versuchst, einen Klassifikator zu erstellen oder das Modell einem Muster folgen zu lassen, stelle sicher, dass genügend Beispiele vorhanden sind. Korrigiere deine Beispiele sorgfältig – das Modell ist normalerweise schlau genug, grundlegende Rechtschreibfehler zu erkennen und dir eine Antwort zu geben, aber es könnte auch annehmen, dass dies absichtlich ist, was die Antwort beeinflussen kann.

**Überprüfe deine Einstellungen.** Die Einstellungen für Temperatur und top_p steuern, wie deterministisch das Modell bei der Generierung einer Antwort ist. Wenn du eine Antwort möchtest, bei der es nur eine richtige Lösung gibt, solltest du diese Werte niedriger einstellen. Wenn du vielfältigere Antworten suchst, möchtest du sie möglicherweise höher einstellen. Der häufigste Fehler bei diesen Einstellungen ist die Annahme, dass sie „Cleverness“ oder „Kreativität“ steuern.

## Einbettungen
Dieser Abschnitt zeigt, wie man Einbettungen abruft und Ähnlichkeiten zwischen Wörtern, Sätzen und Dokumenten findet. Um die folgenden Notebooks auszuführen, müssen Sie ein Modell bereitstellen, das `text-embedding-ada-002` als Basismodell verwendet, und dessen Bereitstellungsnamen in der .env-Datei mit der Variablen `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` festlegen.


### Modell-Taxonomie - Auswahl eines Einbettungsmodells

**Modell-Taxonomie**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (Einbettungsfamilie)  
{capability} --> ada             (alle anderen Einbettungsmodelle werden 2024 eingestellt)  
{input-type} --> n/a             (nur für Suchmodelle angegeben)  
{identifier} --> 002             (Version 002)  

model = 'text-embedding-ada-002'


  > [!HINWEIS] Der folgende Schritt ist nicht erforderlich, wenn dieses Notebook in Codespaces oder innerhalb eines Devcontainers ausgeführt wird


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
