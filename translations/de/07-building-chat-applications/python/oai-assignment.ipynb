### 2. Hilfsbibliotheken importieren und Zugangsdaten instanziieren


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Das richtige Modell finden  
Modelle wie GPT-4o und GPT-4o mini können natürliche Sprache verstehen und generieren und sind auf der OpenAI-Plattform mit verschiedenen Leistungs- und Geschwindigkeitsstufen verfügbar, die für unterschiedliche Aufgaben geeignet sind.


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

## 4. Promptgestaltung  

„Die Magie großer Sprachmodelle besteht darin, dass sie durch das Trainieren zur Minimierung dieses Vorhersagefehlers über riesige Mengen an Text Konzepte lernen, die für diese Vorhersagen nützlich sind. Zum Beispiel lernen sie Konzepte wie“(1):

* wie man buchstabiert
* wie Grammatik funktioniert
* wie man umformuliert
* wie man Fragen beantwortet
* wie man eine Unterhaltung führt
* wie man in vielen Sprachen schreibt
* wie man programmiert
* usw.

#### Wie man ein großes Sprachmodell steuert  
„Von allen Eingaben in ein großes Sprachmodell ist bei weitem die einflussreichste der Textprompt(1).

Große Sprachmodelle können auf einige Arten zur Ausgabe angeregt werden:

Anweisung: Sage dem Modell, was du möchtest
Vervollständigung: Veranlasse das Modell, den Anfang dessen, was du möchtest, zu vervollständigen
Demonstration: Zeige dem Modell, was du möchtest, entweder mit:
Ein paar Beispielen im Prompt
Hunderten oder Tausenden von Beispielen in einem Fine-Tuning-Trainingsdatensatz“



#### Es gibt drei grundlegende Richtlinien für die Erstellung von Prompts:

**Zeigen und sagen**. Mach klar, was du möchtest, entweder durch Anweisungen, Beispiele oder eine Kombination von beidem. Wenn du möchtest, dass das Modell eine Liste von Elementen alphabetisch sortiert oder einen Absatz nach Stimmung klassifiziert, zeige ihm, dass das dein Wunsch ist.

**Biete qualitativ hochwertige Daten an**. Wenn du versuchst, einen Klassifikator zu erstellen oder das Modell ein Muster folgen zu lassen, stelle sicher, dass genügend Beispiele vorhanden sind. Achte darauf, deine Beispiele Korrektur zu lesen – das Modell ist normalerweise intelligent genug, grundlegende Rechtschreibfehler zu durchschauen und dir eine Antwort zu geben, aber es könnte auch annehmen, dass dies beabsichtigt ist, und das kann die Antwort beeinflussen.

**Überprüfe deine Einstellungen.** Die Einstellungen für Temperatur und top_p steuern, wie deterministisch das Modell bei der Generierung einer Antwort ist. Wenn du eine Antwort erfragst, bei der es nur eine richtige Antwort gibt, möchtest du diese niedriger einstellen. Wenn du vielfältigere Antworten möchtest, solltest du sie höher einstellen. Der häufigste Fehler bei diesen Einstellungen ist die Annahme, dass sie die „Klugheit“ oder „Kreativität“ steuern.


Quelle: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Einreichen!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Text zusammenfassen  
#### Herausforderung  
Fasse einen Text zusammen, indem du am Ende eines Textabschnitts ein 'tl;dr:' hinzufügst. Beachte, wie das Modell versteht, wie man eine Reihe von Aufgaben ohne zusätzliche Anweisungen ausführt. Du kannst mit beschreibenderen Eingabeaufforderungen als tl;dr experimentieren, um das Verhalten des Modells zu verändern und die Zusammenfassung anzupassen, die du erhältst(3).  

Jüngste Arbeiten haben bei vielen NLP-Aufgaben und Benchmarks erhebliche Verbesserungen durch Vortraining auf einem großen Textkorpus gefolgt von Feineinstellung auf eine spezifische Aufgabe gezeigt. Während die Architektur typischerweise aufgaben-agnostisch ist, erfordert diese Methode dennoch aufgabenspezifische Feineinstellungsdatensätze mit Tausenden oder Zehntausenden von Beispielen. Im Gegensatz dazu können Menschen in der Regel eine neue Sprachaufgabe mit nur wenigen Beispielen oder einfachen Anweisungen ausführen – etwas, womit aktuelle NLP-Systeme noch weitgehend zu kämpfen haben. Hier zeigen wir, dass die Skalierung von Sprachmodellen die aufgaben-agnostische Few-Shot-Leistung erheblich verbessert und manchmal sogar Wettbewerbsfähigkeit mit früheren State-of-the-Art-Feineinstellungsansätzen erreicht.  



Tl;dr
