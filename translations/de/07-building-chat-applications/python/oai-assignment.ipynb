### Erstellen Sie Ihren ersten Prompt  
Diese kurze Übung bietet eine grundlegende Einführung, wie man Prompts an ein OpenAI-Modell für eine einfache Aufgabe "Zusammenfassung" sendet.


**Schritte**:  
1. Installieren Sie die OpenAI-Bibliothek in Ihrer Python-Umgebung

3. Wählen Sie ein Modell für Ihre Aufgabe  
4. Erstellen Sie einen einfachen Prompt für das Modell  
5. Senden Sie Ihre Anfrage an die Modell-API!


### 1. OpenAI installieren

### 2. Importieren Sie Hilfsbibliotheken und instanziieren Sie Anmeldeinformationen


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Das richtige Modell finden  
Modelle wie GPT-4o und GPT-4o mini können natürliche Sprache verstehen und erzeugen und sind auf der OpenAI-Plattform mit verschiedenen Leistungs- und Geschwindigkeitsebenen erhältlich, die für unterschiedliche Aufgaben geeignet sind.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Prompt-Gestaltung  

„Die Magie großer Sprachmodelle liegt darin, dass sie durch das Trainieren darauf, diesen Vorhersagefehler über enorme Textmengen hinweg zu minimieren, Konzepte lernen, die für diese Vorhersagen nützlich sind. Zum Beispiel lernen sie Konzepte wie“(1):

* wie man buchstabiert
* wie Grammatik funktioniert
* wie man umschreibt
* wie man Fragen beantwortet
* wie man ein Gespräch führt
* wie man in vielen Sprachen schreibt
* wie man programmiert
* usw.

#### Wie man ein großes Sprachmodell steuert  
„Von allen Eingaben in ein großes Sprachmodell ist mit Abstand der Texteingabeaufforderung der einflussreichste Faktor(1).

Große Sprachmodelle können auf verschiedene Weise dazu gebracht werden, Ausgaben zu erzeugen:

Anweisung: Sag dem Modell, was du möchtest
Vervollständigung: Veranlasse das Modell, den Anfang dessen zu vervollständigen, was du möchtest
Demonstration: Zeige dem Modell, was du möchtest, mit entweder:
Ein paar Beispielen in der Eingabeaufforderung
Hunderten oder Tausenden von Beispielen in einem Fine-Tuning-Datensatz“



#### Es gibt drei Grundregeln zur Erstellung von Prompts:

**Zeigen und sagen**. Mach klar, was du möchtest, entweder durch Anweisungen, Beispiele oder eine Kombination aus beidem. Wenn du möchtest, dass das Modell eine Liste von Elementen in alphabetischer Reihenfolge sortiert oder einen Absatz nach Stimmung klassifiziert, zeige ihm, dass das das ist, was du möchtest.

**Qualitätsdaten bereitstellen**. Wenn du versuchst, einen Klassifikator zu bauen oder das Modell einem Muster folgen zu lassen, stelle sicher, dass genügend Beispiele vorhanden sind. Korrigiere deine Beispiele sorgfältig – das Modell ist normalerweise klug genug, um grundlegende Rechtschreibfehler zu erkennen und dir eine Antwort zu geben, aber es könnte auch annehmen, dass dies absichtlich ist, was die Antwort beeinflussen kann.

**Überprüfe deine Einstellungen.** Die Einstellungen „temperature“ und „top_p“ steuern, wie deterministisch das Modell eine Antwort erzeugt. Wenn du eine Antwort möchtest, bei der es nur eine richtige Lösung gibt, solltest du diese Werte niedriger einstellen. Wenn du nach vielfältigeren Antworten suchst, könntest du sie höher einstellen. Der häufigste Fehler, den Menschen hier machen, ist anzunehmen, dass dies Steuerelemente für „Intelligenz“ oder „Kreativität“ sind.


Quelle: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Abschicken!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Text zusammenfassen  
#### Herausforderung  
Fassen Sie Text zusammen, indem Sie am Ende eines Textabschnitts ein „tl;dr:“ hinzufügen. Beachten Sie, wie das Modell versteht, wie es eine Reihe von Aufgaben ohne zusätzliche Anweisungen ausführen kann. Sie können mit beschreibenderen Eingabeaufforderungen als tl;dr experimentieren, um das Verhalten des Modells zu ändern und die Zusammenfassung, die Sie erhalten, anzupassen(3).  

Jüngste Arbeiten haben durch das Vortrainieren an einem großen Textkorpus, gefolgt von der Feinabstimmung auf eine spezifische Aufgabe, erhebliche Fortschritte bei vielen NLP-Aufgaben und Benchmarks gezeigt. Während diese Methode architekturbedingt in der Regel aufgabenunabhängig ist, erfordert sie dennoch aufgabenspezifische Feinabstimmungsdatensätze mit Tausenden oder Zehntausenden von Beispielen. Im Gegensatz dazu können Menschen eine neue Sprachaufgabe in der Regel nur anhand weniger Beispiele oder einfacher Anweisungen ausführen – etwas, womit aktuelle NLP-Systeme größtenteils noch Schwierigkeiten haben. Hier zeigen wir, dass die Vergrößerung von Sprachmodellen die aufgabenunabhängige Few-Shot-Leistung erheblich verbessert und manchmal sogar mit vorherigen State-of-the-Art-Feinabstimmungsansätzen konkurrieren kann.  



Tl;dr
