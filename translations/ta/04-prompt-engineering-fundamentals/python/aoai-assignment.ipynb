# உட்பிரவேசம் சுட்டிகாட்டு பொறியியலில்
சுட்டிகாட்டு பொறியியல் என்பது இயற்கை மொழி செயலாக்க பணிகளுக்கான சுட்டிகளையை வடிவமைக்கும் மற்றும் மிகைப்படுத்தும் செயல்முறை ஆகும். இது சரியான சுட்டிகாட்டுகளை தேர்ந்தெடுக்கும், அவற்றின் அளவுகோல்களை சீர்செய்யும் மற்றும் அவற்றின் செயல்திறனை மதிப்பீடு செய்வதை உட்படுத்துகிறது. சொல் செயலாக்க மாதிரிகளில் உயர் துல்லியத்தையும் செயல்திறனையும் அடைவதில் சுட்டிகாட்டு பொறியியல் மிகவும் முக்கியமானது. இந்த பகுதியில், ஆய்வுக்காக OpenAI மாதிரிகளை பயன்படுத்தி சுட்டிகாட்டு பொறியியலின் அடிப்படைகளை நாம் ஆராயப்போகிறோம்.


### பயிற்சி 1: டோக்கனீசேஷன்
OpenAIவில் இருந்து ஒரு திறந்த மூல அதிவேக டோக்கனையசர் olan tiktoken ஐ பயன்படுத்தி டோக்கனீசேஷனை ஆராயுங்கள்
மேலும் உதாரணங்களுக்கு [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) காணவும்.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### பயிற்சி 2: OpenAI API விசை அமைப்பை சரிபார்க்கவும்

கீழே உள்ள குறியீட்டை ஓட்டி உங்கள் OpenAI என்ட்பாயின்ட் சரியாக அமைக்கப்பட்டதா என்பதை சரிபார்க்கவும். குறியீடு எளிய அடிப்படையிலான ஒரு கோரிக்கையை முயற்சித்து முடித்தலை சரிபார்க்கின்றது. உள்ளீடு `oh say can you see` என்பது `by the dawn's early light..` என்ற வரிகளுக்கு பூர்த்தியாக வேண்டும்.


````python
# Uses the OpenAI client against the Azure OpenAI (Microsoft Foundry) v1 endpoint
# with the Responses API. See https://aka.ms/openai/start

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### பயிற்சி 3: வேதனைகள்
ஒரு தலைப்பைப் பற்றி LLM-க்கு ஒரு கேள்வியை முன்வைத்து அதைத் தெரியாமலோ அல்லது அதன் முன்னேற்பாடு செய்யப்பட்ட தரவுத் தொகுப்புக்கு வெளியிலோ (சமீபத்திய) இருக்கக்கூடிய தலைப்புகளுக்கு பற்றிய நிறைவேற்றல்களை திரும்ப பெறும்போது என்ன நடக்கும் என்பதை ஆராயுங்கள். நீங்கள் வேறு ஒரு கேள்வியை, அல்லது வேறு ஒரு மாதிரியை முயற்பித்தால் பதில் எப்படி மாறுகிறது என்பதையும் பாருங்கள்.


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### பயிற்சி 4: கோரிக்கை அடிப்படையிலான 
முதன்மை உள்ளடக்கத்தை அமைக்க "text" மாறியைப் பயன் படுத்தவும்
மற்றும் அந்த முதன்மை உள்ளடக்கத்துடன் தொடர்புடைய வழிமுறையை வழங்க "prompt" மாறியை பயன்படுத்தவும்.

இங்கு நாங்கள் மாதிரியை இரண்டாம் வகுப்பு மாணவருக்காக உரையை சுருக்கச் சொல்லுகிறோம்


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### பயிற்சி 5: சிக்கலான ப்ராம்ட்  
அமைப்பு, பயனர் மற்றும் உதவியாளர் செய்திகளை கொண்ட ஒரு கோரிக்கையை முயற்சிக்கவும்  
அமைப்பு உதவியாளர் சூழலை அமைக்கிறது  
பயனர் மற்றும் உதவியாளர் செய்திகள் பல முறை உரையாடல் சூழலை வழங்குகின்றன  

உதவியாளர் குணாதிசயத்தை அமைப்பில் "எரிச்சலாக" அமைக்கப்பட்டுள்ளதை கவனிக்கவும்.  
வேறு குணாதிசய சூழலை பயன்படுத்தி முயற்சிக்கவும். அல்லது வேறு உள்ளீடு/வெளியீடு செய்திகளின் தொடர் ஒன்றையும் முயற்சிக்கவும்  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
