## DALL-E மற்றும் Midjourney எப்படி செயல்படுகின்றன?

ஒரு *ஆட்டோரோட்டகிரஸ்வ் டிரான்ஸ்ம்வர்மர்* என்பது மாதிரி எப்படி உரை விளக்கத்திலிருந்து படங்களை உருவாக்குகிறது என்பதைக் குறிக்கிறது, இது ஒவ்வொரு pixel-ஐ ஒரு நேரத்தில் உருவாக்கி, அடுத்த pixel-ஐ உருவாக்க அதே pixel-கள் பயன்படுத்துவதாகும். நியூரல் நெட்வொர்க்கின் பல அடுக்குகளுக்குள் பயணித்து, படம் முடிவடையும் வரை.

இந்த செயல்முறையால், DALL-E உருவாக்கும் படத்தில் பண்புகள், பொருட்கள், குணாதிசயங்கள் மற்றும் மேலும் பலவற்றை கட்டுப்படுத்துகிறது. ஆனால், DALL-E 2 மற்றும் 3 கள் உருவாக்கப்பட்ட படத்தை மீண்டும் மேலான கட்டுப்பாடு வைத்துள்ளன,


## உங்கள் முதல் பட உருவாக்க பயன்பாட்டை கட்டமைக்கிறது

ஆகவே ஒரு பட உருவாக்க பயன்பாட்டை கட்டமைக்க என்ன வேண்டும்? நீங்கள் கீழ்க்கண்ட நூலகங்கள் தேவை:

- **python-dotenv**, உங்கள் இரகசியங்களை *.env* கோப்பில் கோடுகளிலிருந்து தனித்து வைக்க இந்த நூலகத்தை பயன்படுத்த பரிந்துரைக்கப்படுகிறீர்கள்.
- **openai**, இந்த நூலகம் OpenAI API உடன் தொடர்பு கொள்ள நீங்கள் பயன்படுத்தப் போகும்.
- **pillow**, Python இல் படங்களுடன் வேலை செய்ய.
- **requests**, HTTP கோரிக்கைகளைச் செய்ய உதவ.

இன்னும் செய்யாவிட்டால், [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) பக்கத்தில் உள்ள வழிமுறைகளை பின்பற்றவும் Azure OpenAI வளம் மற்றும் மாதிரியை உருவாக்க. மாதிரியாக **gpt-image-1** ஐத் தேர்ந்தெடுக்கவும் (தற்போதைய தலைமுறை Azure OpenAI பட மாதிரி; DALL-E 3 பழையது மற்றும் புதிய நிறுவல்களுக்கு கிடைக்காது).

1. பின்வரும் உள்ளடக்கத்துடன் *.env* என்ற கோப்பை உருவாக்கவும்:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    இந்த தகவலை உங்கள் வளத்தின் மைக்ரோசாஃப்ட் ஃபௌன்ட்ரி போர்டல் "Deployments" பகுதியில்மேல் கண்டறியவும்.



1. மேலே உள்ள நூலகங்களை *requirements.txt* என்ற கோப்பில் கீழ் காணப்படும் விதமாக சேகரிக்கவும்:

    ```text
    python-dotenv
    openai

    ```


1. அடுத்து, ஒரு மெய்நிகர் சூழலை உருவாக்கி நூலகங்களை நிறுவுக:

1. *app.py* எனும் கோப்பில் கீழ்கண்ட குறியீட்டை சேர்க்கவும்:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv ஐ இறக்குமதி செய்யவும்
    dotenv.load_dotenv()

    # Azure OpenAI சேவை கிளையன்டை அமைக்கவும்
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # படம் உருவாக்க API ஐ பயன்படுத்தி ஒரு படம் உருவாக்கவும்
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் வேண்டுகோள் உரையை இங்கே உள்ளிடவும்
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # சேமிக்கப்பட்ட படத்துக்கான அடைவைப் அமைக்கவும்
        image_dir = os.path.join(os.curdir, 'images')
    
        # அடைவு ஏதுவாக இல்லாவிட்டால், அதை உருவாக்கவும்
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # படப் பாதையை துவங்கவும் (கோப்புருவம் png ஆக இருக்க வேண்டும்)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # உருவாக்கப்பட்ட படத்தைப் பெறவும்
        image_url = generation_response.data[0].url  # பதிலிலிருந்து பட URL ஐ எடுக்கவும்
        generated_image = requests.get(image_url).content  # படத்தை பதிவிறக்கம் செய்யவும்
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # இயல்புநிலை பட பார்வையில் படத்தை காட்டவும்
        image = Image.open(image_path)
        image.show()
    
    # தவிர்க்கப்பட்ட பிழைகளை பிடிக்கவும்
    except openai.BadRequestError as err:
        print(err)

    ```

- முதலில், நமக்கு தேவைப்படும் நூலகங்களை இறக்குமதி செய்கிறோம், அதில் OpenAI நூலகம், dotenv நூலகம், requests நூலகம் மற்றும் Pillow நூலகம் அடங்கும்.

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    ```

- பிறகு, *.env* கோப்பிலிருந்து சுற்றுச்சூழல் மாறிலிகளை ஏற்றுகிறோம்.

    ```python
    # dotenv ஐ இறக்குமதி செய்க
    dotenv.load_dotenv()
    ```

- அதன்பிறகு, Azure OpenAI சேவை கிளையை கட்டமைக்கிறோம்.

    ```python
    # Azure OpenAI சேவை கிளையண்டை அமைக்கு
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
    ```

- அடுத்து, படத்தை உருவாக்குகிறோம்:

    ```python
    # பட உருவாக்கும் APIயைப் பயன்படுத்தி ஒரு படம் உருவாக்கவும்
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் ஊக்கவிருப்ப உரையை இங்கே உள்ளிடவும்
    ```

    ```python
    image = Image.open(image_path)
    ```

### படத்தை உருவாக்குவத بارے விரிவான தகவல்கள்

படத்தை உருவாக்கும் குறியீட்டை விரிவாக பார்க்கலாம்:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் செயற்பாட்டு உரையை இங்கே உள்ளிடவும்
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, உருவாக்க படத்திற்கு பயன்படுத்தப்படும் உரை உத்வேகம். இந்த வேளையில், நாங்கள் "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" என்ற உத்வேகத்தை பயன்படுத்துகிறோம்.
- **size**, உருவாக்கப்படும் படத்தின் அளவு. இந்த இடத்தில், 1024x1024 பிக்சல்கள் அளவுடன் படம் உருவாக்கப்படுகிறது.
- **n**, உருவாக்கப்படும் படங்களின் எண்ணிக்கை. இங்கு, ஒரு படம் மட்டுமே உருவாக்கப்படுகிறது.

படங்களுடன் நீங்கள் செய்யக்கூடிய கூடுதல் செயல்கள் அடுத்த பகுதியில் பார்க்கப்படும்.

## பட உருவாக்கத்தின் கூடுதல் திறன்கள்

இங்கே வரையில், Python-இல் சில வரிகளால் படத்தை உருவாக்க முடியுமென பார்த்துள்ளீர்கள். ஆனால் படம் தொடர்பான கூடுதல் செயல்கள் மேற்கொள்ள முடியும்.

நீங்கள் பின்வருமாறு செய்யலாம்:

- **திருத்தங்கள் செய்யவும்**. ஏற்கனவே உள்ள படத்திற்கு ஒரு மாதிரியை மற்றும் உத்வேகத்தை வழங்கி, படத்தை மாற்றலாம். உதாரணமாக, ஒரு படத்தின் ஒரு பகுதியை மாற்றி அதில் என்னை சேர்க்கலாம். நமது முயலைப் படம் எடுத்துக்கொண்டால், முயலுக்கு தொப்பியை சேர்க்கலாம். அதை செய்வது எப்படி என்றால், படம், மாற்றம் செய்யவேண்டிய பகுதிக்கு உள்ளது என்று காட்டும் மாதிரி, மற்றும் என்ன செய்யவேண்டும் என்கிறது குறிக்கும் உரை உத்வேகத்தை வழங்குவது.

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    அடிப்படை படம் முயல் மட்டுமே கொண்டிருக்கும், ஆனால் இறுதி படத்தில் முயலுக்கு தொப்பி இருக்கும।
