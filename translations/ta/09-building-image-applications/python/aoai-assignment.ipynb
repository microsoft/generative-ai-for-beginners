# பட உருவாக்க பயன்பாடுகளை உருவாக்குதல் (அழூர் OpenAI)

LLM-களுக்கு உரை உருவாக்கத்தின் அப்பால் இன்னும் பலவீனங்கள் உள்ளன. உரை விளக்கங்களிலிருந்து படங்களை உருவாக்கவும் முடியும். படங்கள் ஒரு முறைமையாக மருத்துவக் தொழில்நுட்பம், கட்டிடக்கலை, சுற்றுலா, விளையாட்டு வளர்ப்பு, சந்தைப்படுத்தல் மற்றும் பலல் துறைகளில் பயன்படுகிறது. இந்த பாடத்தில் நாம் இன்றைய **GPT Image** மாதிரிகளை Microsoft Foundry-ல் பயன்படுத்துகின்றோம்.

## கற்றல் குறிக்கோள்கள்

இந்த பாடம் முடிந்தவுடன் நீங்கள் செய்யக்கூடியவை:

- பட உருவாக்கம் என்ன மற்றும் அது எங்கே பயன்படுகிறது என்பதை விளக்குங்கள்.
- `gpt-image` மாதிரி குடும்பத்தை அறிவது மற்றும் அது பாரம்பரிய DALL·E மாதிரிகளிலிருந்து எப்படி வேறுபடுகிறது என்பதை புரிந்து கொள்ளுங்கள்.
- ஒரு பட உருவாக்க பயன்பாட்டை கட்டியமைக்கவும் படங்களை திருத்தவும்.

## பட உருவாக்கம் என்றால் என்ன?

பட உருவாக்க மாதிரிகள் ஒரு உரை முன்மொழிவிலிருந்து படங்களை உருவாக்குகின்றன. `gpt-image` போன்ற நவீன மாதிரிகள் பயிற்சியின் போது உரை மற்றும் படங்களுக்கிடையேயான உறவினை கற்றுக்கொள்கின்றன, பின்னர் முறையாக சரவான ரேண்டம் சத்தத்தை படமாக மாற்றுகின்றன, அது உங்கள் முன்மொழிவுக்கு பொருந்தும்.

இரண்டு சிறந்த பரிசுருதி பெற்ற பட மாதிரி குடும்பங்கள் அவை:

- **`gpt-image` (OpenAI)** — இந்த பாடத்தில் பயன்படுத்தப்படும் தற்போதைய தலைமுறை. இது உரை-பட உருவாக்கம் மற்றும் பட திருத்துதலை (மாசு கொண்டு inpainting) ஆதரிக்கின்றது.
- **Midjourney** — தனக்கான சேவை மற்றும் Discord அடிப்படையிலான வேலைதிட்டம் உடைய பிரபல மூன்றாம் தரப்பின் மாதிரி.

> பழைய OpenAI பட மாதிரிகள் — **DALL·E 2** மற்றும் **DALL·E 3** — பாரம்பரியமானவை. DALL·E 3 இப்போது புதிய அம்சங்களுக்கு கிடைக்கவில்லை, மேலும் `create_variation` போன்ற அம்சங்கள் DALL·E 2யில் மட்டுமே இருந்தன. புதிய பயன்பாடுகளுக்கு `gpt-image` மாதிரிகளை பயன்படுத்துங்கள்.

Microsoft Foundry-ல், **`gpt-image-2`** சமீபத்திய மற்றும் மிக திறமையான பட மாதிரியாகும் மற்றும் பரிந்துரைக்கப்படும் இயல்பானது. `gpt-image-1.5` மற்றும் `gpt-image-1-mini` கூட பொதுவாக கிடைக்கின்றன.

> **முக்கியம்:** `gpt-image` மாதிரிகள் உருவாக்கப்பட்ட படத்தை **base64** (`b64_json`) என்று தருகின்றன, URL அல்ல. உங்கள் கோடு base64 ம stringய்யை பைட்களாக டிக்கோடு செய்து சேமிக்கிறது — பதிவிறக்கம் செய்ய பட URL கிடையாது.


## உங்கள் முதல் படம் உருவாக்கும் செயலியை உருவாக்கல்

அப்படியானால், ஒரு படம் உருவாக்கும் செயலியை உருவாக்க என்ன என்ன தேவை? உங்கள் தேவைகள் கீழ்காணும் நூலகங்களை உள்ளடக்கியவையாக இருக்க வேண்டும்:

- **python-dotenv**, உங்கள் ரகசியங்களை *.env* கோப்பில் வைத்திருப்பதற்காக இந்த நூலகத்தை பயன்படுத்த பரிந்துரைக்கப்படுகிறது.
- **openai**, OpenAI API உடன் தொடர்பு கொள்ள நீங்கள் இந்த நூலகத்தை பயன்படுத்த வேண்டும்.
- **pillow**, Python இல் படங்களை வேலை செய்ய.

இதுவரை செய்யப்படவில்லையெனில், Microsoft Foundry வளம் மற்றும் மாதிரியை உருவாக்க [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) பக்கத்தில் உள்ள அறிவுரைகளை பின்பற்றவும். மாதிரியாக **gpt-image-2** ஐத் தேர்ந்தெடுக்கவும் (கடைசியாக உள்ள Azure OpenAI படம் மாதிரி; DALL·E பழையது).

1. கீழ்காணும் உள்ளடக்கத்துடன் *.env* என்ற கோப்பை உருவாக்கவும்:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    இந்த தகவல்களை உங்கள் வளத்திற்கு Microsoft Foundry போர்டலில் "Deployments" பகுதியில் கண்டறியலாம்.



1. மேலே குறிப்பிடப்பட்ட நூலகங்களை *requirements.txt* என்ற கோப்பில் இதுபோல் சேகரிக்கவும்:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. அடுத்ததாக, மெய்ப்பாணச் சூழலை உருவாக்கி நூலகங்களை நிறுவவும்:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> விண்டோஸுக்காக, உங்கள் مجازی சூழலை உருவாக்கி செயல்படுத்த கீழ்கண்ட கட்டளைகளைப் பயன்படுத்தவும்:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* என்ற கோப்பில் பின்வரும் குறியீட்டைச் சேர்க்கவும்:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv ஐ இறக்குமதி செய்க
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) கிளையண்டை அமைக்கவும்.
    # பட முகாமodelகளுக்கு சமீபத்திய API பதிப்பு தேவை — உங்கள் மாடலுக்கு தேவையான பதிப்பை Foundry ஆவணங்களில் பார்த்துகொள்ளவும்.
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # படம் உருவாக்க பட உருவாக்க API ஐப் பயன்படுத்தவும்
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் ஊக்க உரையை இங்கே உள்ளிடவும்
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # உதாரணம்: "gpt-image-2"
        )
        # சேமிக்கப்பட்ட படத்திற்கான அடைவை அமைக்கவும்
        image_dir = os.path.join(os.curdir, 'images')

        # அந்த அடைவு இல்லாவிடில், அதனை உருவாக்கவும்
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # பட பாதையைக் தொடக்கம் செய்க (கோப்பு வகை png ஆக இருக்க வேண்டும்)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image மாடல்கள் படத்தை URL அல்லாமல் base64 (b64_json) ஆக வழங்கும்
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # படத்தை இயல்புநிலை படக் காண்பிப்பான் மூலம் காட்டு
        image = Image.open(image_path)
        image.show()

    # தவறுகளை பறிமாற்றவும்
    except BadRequestError as err:
        print(err)

    ```

இந்த குறியீட்டைப் பற்றி விளக்குவோம்:

- முதலில், நாம் தேவையான நூலகங்களை ஏற்றுகிறோம், அதாவது OpenAI நூலகம், dotenv நூலகம், base64 மொடியூல் மற்றும் Pillow நூலகம்.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- அதன் பிறகு, *.env* கோப்பிலிருந்து சுற்றுச்சூழல் மாறிலிகள் ஏற்றப்படுகிறது.

    ```python
    # dotenvஐ இறக்குமதி செய்க
    dotenv.load_dotenv()
    ```

- அதன்பிறகு, Azure OpenAI (Microsoft Foundry) கிளையண்ட் உள்ளமைக்கப்படுகிறது.

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- அடுத்ததாக, படத்தை உருவாக்குகிறோம்:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் உரையின் உரையை இங்கே உள்ளிடுக
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` மாதிரிகள் படத்தை **base64** தொடர் வடிவில் `data[0].b64_json` இலிருந்து தருகின்றன. அதை பைட்கள் ஆக குறியாக்கி ஒரு கோப்பில் எழுதுகிறோம் — பதிவிறக்கம் செய்ய URL இல்லை.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- கடைசியாக, படத்தைத் திறந்துவிட்டு சாதாரண பட பார்வையாளரைப் பயன்படுத்தி காட்டுகிறோம்:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### படத்தை உருவாக்குவதற்கான கூடுதல் விவரங்கள்

`images.generate` என்ற செயலியின் அளவுருக்களைப் பார்ப்போம்:

- **prompt**, படம் உருவாக்க பயன்படுத்தபடும் உரை ஊக்கம். இங்கு அது "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, உருவாக்கப்படும் படத்தின் விவரம் (எ.கா. `1024x1024`, `1536x1024`, `1024x1536`, அல்லது `"auto"`).
- **n**, உருவாக்கப்படும் படங்களின் எண்ணிக்கை. இங்கு ஒரு படம் உருவாக்கப்படுகிறது.
- **model**, உங்கள் பட மாதிரி வரம்பு பெயர் (எ.கா. `gpt-image-2`).

> படம் மாதிரிகள் `temperature` என்ற அளவுருவை எடுத்துக்கொள்ளாது — அது உரை உருவாக்க கட்டுப்பாடு ஆகும். மாறுபாட்டுக்கு, API ஐ மீண்டும் அழைக்கவும்; மாறுபாட்டை குறைக்க, உங்கள் ஊக்கத்தை குறிப்பாக உருவாக்கவும்.

## பட உருவாக்கத்தின் கூடுதல் திறன்கள்

நீங்கள் Python-இல் சில வரிகள் மூலம் படத்தை எப்படி உருவாக்குவது என்று பார்த்தீர்கள். `gpt-image` மாதிரிகள் ஒரு இருந்த படத்தையும் **தேர்ந்தெடுத்து திருத்த** முடியும். ஒரு படம், விருப்பமான **மாஸ்க்** (மாற்றுவுலக பகுதியில் குறியிடுக), மற்றும் ஒரு ஊக்கம் வழங்கி,ப் படத்தின் ஒரு பகுதியை மாற்ற முடியும் — உதாரணமாக, நமது கழுதைக்கு தொப்பி சேர்க்கலாம்.

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# மாற்றங்களும் base64 ஆகவும் திருப்பிக் கொடுக்கப்படுகின்றன
edited_image = base64.b64decode(response.data[0].b64_json)
```

அடிப்படை படம் சிங்கத்தை மட்டுமே கொண்டுள்ளது; இறுதி படத்தில் கழுதைக்கு தொப்பி இருக்கிறது.
