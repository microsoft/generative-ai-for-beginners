# பட விளையாட்டு உருவாக்கும் செயலிகள் (OpenAI)

LLMகளுக்கு உரைonly உருவாக்கம் மட்டுமல்ல. நீங்கள் உரை விளக்கங்களிலிருந்து படங்களைக் கொடுக்கவும் முடியும். படங்கள் ஒரு முறைமை ஆகி மருத்துவ தொழில்நுட்பம், கட்டிடக்கலை, சுற்றுலா, கேம் மேம்பாடு, சந்தைப்படுத்தல் மற்றும் இன்னும் பல துறைகளில் பயன்படுகிறது. இந்த பாடத்தில் நாம் இன்று உள்ள **GPT பட** மாதிரிகளுடன் OpenAI தளத்தில் வேலை செய்கிறோம்.

## கற்றல் குறிக்கோள்கள்

இந்த பாடத்தின் முடிவில் நீங்கள்:

- பட விளையாட்டு என்றால் என்ன மற்றும் அது எங்கே பயனுள்ளதாக இருக்கிறதென விளக்க முடியும்.
- `gpt-image` மாதிரி குடும்பத்தை மற்றும் அது புறநிலை DALL·E மாதிரிகளிலிருந்து எப்படி வேறுபடுகின்றது என்பதை புரிந்து கொள்வது.
- ஒரு பட விளையாட்டு செயலியை உருவாக்கி படங்களை தொகுக்க முடியும்.

## பட விளையாட்டு என்றால் என்ன?

படம் உருவாக்கும் மாதிரிகள் உரை பிறட்டை கொண்டு படங்களை உருவாக்குகின்றன. `gpt-image` மாதிரிகளின் போன்ற நவீன மாதிரிகள் பயிற்சி காலையில் உரை மற்றும் படம் இடையேயான தொடர்பை கற்றுக்கொள்கின்றன, பின்னர் இடையில் உள்ள சீரற்ற சத்தை படமாக மாறச் செயல்படுகின்றன.

இரண்டு பிரபலமான படம் மாதிரி குடும்பங்கள்:

- **`gpt-image` (OpenAI)** — இந்த பாடத்தில் நாம் பயன்படுத்தும் தற்போதைய தலைமுறை மாதிரி. இது உரை முதல் படம் உருவாக்கும் மற்றும் பட தொகுத்தல் (மாஸ்க் மூலம் உள்ளீடு) முறைகளை ஆதரிக்கிறது.
- **Midjourney** — தனிப்பட்ட சேவையுடன் மற்றும் Discord அடிப்படையிலான வேலை வழங்கும் பரபரப்பான மூன்றாம் கட்சி மாதிரி.

> பழைய OpenAI படம் மாதிரிகள் — **DALL·E 2** மற்றும் **DALL·E 3** — வழக்கமானவை. DALL·E 3 புதிய பயன்படுத்துகளுக்கு கிடைக்கவில்லை, மற்றும் `create_variation` போன்ற அம்சங்கள் DALL·E 2க்குள் மட்டுமே இருந்தன. புதிய செயலிகளுக்கு `gpt-image` மாதிரிகளை பயன்படுத்தவும்.

> **முக்கியம்:** `gpt-image` மாதிரிகள் உருவாக்கிய படத்தை **base64** (`b64_json`) வடிவத்தில் தருகின்றன, URL ஆக அல்ல. உங்கள் குறியீடு base64 தொடரை பைட் ஆக மாற்றி சேமிக்கும் — அதனை பதிவிறக்கம் செய்ய பட URL கிடையாது.


## உங்கள் முதல் படம் உருவாக்கும் செயலியில் கட்டமைத்தல்

ஆகவே படம் உருவாக்கும் செயலியை கட்டமைப்பதற்கு என்ன தேவை? நீங்கள் பின்வரும் நூலகங்களைப் பயன்படுத்த வேண்டும்:

- **python-dotenv**, உங்கள் ரகசியங்களை குறியீட்டிலிருந்து விலகி *.env* கோப்பில் சேமிக்க இந்த நூலகத்தை பயன்படுத்த பரிந்துரைக்கப்படுகிறது.
- **openai**, இந்த நூலகம் OpenAI API உடன் தொடர்பு கொள்வதற்காக உபயோகிக்கப்படும்.
- **pillow**, Python இல் படங்களுடன் பணியாற்ற.


1. பின்வரும் உள்ளடக்கத்துடன் *.env* என்ற கோப்பை உருவாக்கவும்:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. மேலே குறிப்பிட்ட நூலகங்களை *requirements.txt* என்ற கோப்பில் கீழ்காணும் படி சேகரிக்கவும்:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. அடுத்தது, மெய்நிகர் சூழலை உருவாக்கி நூலகங்களை நிறுவவும்:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> விண்டோஸுக்காக, உங்கள் விர்ச்சுவல் சூழலை உருவாக்கி செயல்படுத்த கீழ்காணும் கட்டளைகளை பயன்படுத்தவும்:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* என்ற கோப்பில் கீழ்க்காணும் குறியீட்டை சேர்க்கவும்:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv ஐ இறக்குமதி செய்க
    dotenv.load_dotenv()

    # OpenAI பொருளை உருவாக்குக (.env இலிருந்து OPENAI_API_KEY ஐ வாசிக்கிறது)
    client = openai.OpenAI()


    try:
        # பட உருவாக்க API மூலம் ஒரு படம் உருவாக்குக
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் கூற்று உரையை இங்கே உள்ளிடுக
            size='1024x1024',
            n=1
        )
        # சேமிக்கப்பட்ட படத்துக்கான அடைவை அமைக்கவும்
        image_dir = os.path.join(os.curdir, 'images')

        # அடைவு இல்லாவிட்டால், அதனை உருவாக்குக
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # பட பாதையை ஆரம்பிக்கவும் (கோப்பு வகை png ஆக இருக்க வேண்டும்)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image மாடல்கள் படத்தை base64 (b64_json) ஆக வழங்குகின்றன, URL அல்ல
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # இயல்புநிலை படம் பார்வையாளியில் படத்தை காட்டு
        image = Image.open(image_path)
        image.show()

    # தவறுகளை கைப்பற்றுக
    except openai.BadRequestError as err:
        print(err)

    ```

இந்த குறியீட்டை விளக்கமாகப் பார்க்கலாம்:

- முதலில், நம்மிடம் தேவைப்படும் நூலகங்களை இறக்குமதி செய்கிறோம், இதில் OpenAI நூலகம், dotenv நூலகம், base64 மைட்ச் மற்றும் Pillow நூலகம் அடங்கும்.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- அதன்பிறகு, client ஐ உருவாக்குகிறோம், இது உங்கள் ``.env`` இலிருந்து API விசையை வாசிக்கிறது.

    ```python
    # OpenAI பொருள் உருவாக்கவும்
    client = openai.OpenAI()
    ```

- அடுத்து, நாம் படம் உருவாக்குகிறோம்:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் கேள்வி உரையை இங்கே உள்ளிடவும்
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` மாதிரிகள் படம் **base64** நெறிமுறையில் `data[0].b64_json` ஆக தருகின்றன. அதை பைட்டுகளாக்கிப் பெற்று கோப்பாக எழுதுகிறோம் — பதிவிறக்க URL எதுவும் கிடையாது.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- கடைசியாக, படத்தை திறந்து, படத்தைக் காண்பிக்கும் வழக்கமான படம் பார்க்கும் கருவியை பயன்படுத்துகிறோம்:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### படத்தை உருவாக்குவதற்கான மேலும் விவரங்கள்

`images.generate` இன் அளவுருக்கள் பற்றி பார்ப்போம்:

- **model**, பயன்படுத்தும் படம் மாதிரி (எடுத்துக்காட்டாக `gpt-image-1`).
- **prompt**, படத்தை உருவாக்குவதற்கான உரை விளக்கம். இங்கே "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, உருவாக்கப்படுகிற படத்தின் அளவு (எடுத்துக்காட்டாக `1024x1024`, `1536x1024`, `1024x1536`, அல்லது `"auto"`).
- **n**, உருவாக்கபட வேண்டிய படங்கள் எண்ணிக்கை. இங்கே ஒன்று உருவாக்குகிறோம்.

> படம் மாதிரிகள் `temperature` என்ற அளவுருவை எடுத்துக் கொள்ளாது — அது உரை-உருவாக்கக் கட்டுப்பாடு. வித்தியாசம் வேண்டுமானால் மீண்டும் API ஐ அழைக்கவும்; வித்தியாசம் குறைக்க உங்கள் prompt ஐ கட்டாயமாக்கவும்.

## பட உருவாக்கத்தின் கூடுதல் திறன்கள்

சில வரிகளான Python நிரலால் படத்தை உருவாக்குவது நீங்கள் பார்த்திருக்கிறீர்கள். `gpt-image` மாதிரிகள் உள்ள படத்தை **தாயார்** செய்யவும் முடியும். ஒரு படம், விரும்பப்படும் **mask** (மாற்றவேண்டிய பகுதியை குறிக்கும்) மற்றும் விளக்கத்துடன், படத்தின் ஒரு பகுதியை மாற்றலாம் — உதாரணமாக, நம் கூரையுள்ள பட்டாம்பூச்சிக்கு இடைத்துண்டை சேர்க்கலாம்.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# திருத்தங்களும் base64 ஆகவும் திருப்பிக் கொடுக்கப்படுகின்றன
edited_image = base64.b64decode(response.data[0].b64_json)
```

அடிப்படை படத்தில் வெறும் முயல் மட்டுமே உள்ளது; இறுதி படத்தில் முயலைக் கூரையுடன் காணலாம்.
