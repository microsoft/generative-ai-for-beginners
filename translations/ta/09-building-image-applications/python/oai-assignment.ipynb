## உங்கள் முதலாவது பட உருவாக்க பயன்பாட்டை உருவாக்கல்

ஆகவே, பட உருவாக்கும் பயன்பாட்டை உருவாக்க என்ன தேவை? உங்களுக்கு பின்வரும் நூலகங்கள் தேவை:

- **python-dotenv**, உங்கள் ரகசியங்களை குறியீட்டில் இருந்து தனியாக *.env* கோப்பில் வைத்திருக்க இந்த நூலகத்தை பயன்படுத்த பரிந்துரைக்கப்படுகிறது.
- **openai**, இந்த நூலகத்தை OpenAI API உடன் தொடர்பு கொள்ள பயன்படுத்துவீர்கள்.
- **pillow**, பைத்தானில் படங்களுடன் வேலை செய்ய.
- **requests**, HTTP கோரிக்கைகளை செய்ய உதவ.


1. பின்வரும் உள்ளடக்கத்துடன் *.env* ஒரு கோப்பை உருவாக்கவும்:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. மேலே காட்டப்பட்ட நூலகங்களை *requirements.txt* என்ற கோப்பில் கீழேபடி சேகரியுங்கள்:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. பின்பு, விர்ச்சுவல் என்வையிரான்மென்டைப் பிறக்க செய்து நூலகங்களை நிறுவுங்கள்:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> விண்டோசுக்கு, உங்கள் வெர்ச்சுவல் சுற்றுப்பயர்வை உருவாக்கவும் மற்றும் செயல்படுத்த கீழ்காணும் கட்டளைகளை பயன்படுத்தவும்:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* என்ற கோப்பில் கீழ்காணும் குறியீட்டை சேர்க்கவும்:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ஐ இறக்குமதி செய்
    dotenv.load_dotenv()

    # OpenAI பொருளை உருவாக்கவும்
    client = openai.OpenAI()


    try:
        # படம் உருவாக்கும் API ஐ பயன்படுத்தி படம் உருவாக்கவும்
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் வாக்கியத்தை இங்கே உள்ளிடவும்
            size='1024x1024',
            n=1
        )
        # சேமிக்கப்பட்ட படத்திற்கு அடைவகத்தை அமைக்கவும்
        image_dir = os.path.join(os.curdir, 'images')

        # அடைவகம் இல்லையெனில், அதை உருவாக்கவும்
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # பட பாதையை ஆரம்பிக்கவும் (கோப்புப் பாணி png ஆக இருக்க வேண்டும் என்பதை கவனிக்கவும்)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # உருவாக்கப்பட்ட படத்தை பெறுக
        print(generation_response)

        image_url = generation_response.data[0].url # பதிலில் இருந்து பட URL ஐ பிரிக்கவும்
        generated_image = requests.get(image_url).content  # படத்தை பதிவிறக்கம் செய்க
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # இயல்புநிலை பட பார்வையாளரில் படத்தை காட்டுக
        image = Image.open(image_path)
        image.show()

    # தவறுகளை பிடிக்கவும்
    except openai.BadRequestError as err:
        print(err)

    ```

    மேல் குறியீடு உருவாக்கப்பட்ட படத்தின் URL உள்ள JSON பொருளை தருகிறது. அந்த URL ஐப் பயன்படுத்தி படத்தை பதிவிறக்கி கோப்பாக சேமிக்க முடியும்.

- கடைசியாக, படத்தை திறந்து ஸ்டாண்ட் பட பார்வையாளர் மூலம் காட்டுகிறோம்:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### படத்தை உருவாக்கும் வழிமுறைகள் பற்றி கூடுதல் விவரங்கள்

படத்தை உருவாக்கும் குறியீட்டை நன்கு பார்ப்போம்:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # உங்கள் முன்னுரைப் உரையை இங்கே உள்ளிடவும்
        size='1024x1024',
        n=1
    )
```

- **prompt**, என்பது படத்தை உருவாக்க பயன்படும் எழுத்து வழிகாட்டி. இந்த உதாரணத்தில், "மாட்டின் மீது கூப்பல், லாலிப்பாப்பை பிடித்து, மலர் வளங்கி வளரும் மன்றத்தில்" எனும் வழிகாட்டி பயன்படுகிறது.
- **size**, என்பது உருவாக்கப்படும் படத்தின் அளவு. இந்த உதாரணத்தில், படத்தின் அளவு 1024x1024 பிக்சல்கள் ஆகும்.
- **n**, என்பது உருவாக்கப்படும் படங்களின் எண்ணிக்கை. இந்த உதாரணத்தில், இரண்டு படங்கள் உருவாக்கப்படுகின்றன.

படங்களுடன் நீங்கள் செய்யக்கூடிய மேலும் பல விஷயங்கள் அடுத்த பகுதியிலுள்ளோம்.

## பட உருவாக்கத்தின் கூடுதல் திறன்கள்

இதுவரை சில Python வரிகளைக் கொண்டு படத்தை உருவாக்கியவாறு பார்த்தீர்கள். ஆனால் படங்களுடன் செய்யக்கூடிய மற்ற விஷயங்களும் உள்ளன.

நீங்கள் மேலும் பின்வரும்விதமாக செயல்படலாம்:

- **திருத்தங்களைச் செய்யவும்**. ஒரு இருப்பு படத்துடன் ஒரு முகமூடி மற்றும் ஒரு வழிகாட்டியை வழங்குவதனால், படத்தை மாற்ற முடியும். உதாரணமாக, ஒரு ப Bunny படம் எடுத்துக் கொண்டால், அதன் மீது ஒரு தொப்பியைச் சேர்க்கலாம். அதற்காக, படம், மாற்றப்பட வேண்டிய பகுதியைக் குறிப்பிடும் முகமூடி மற்றும் செய்யவேண்டியதைச் சொல்லும் எழுத்து வழிகாட்டி வழங்க வேண்டும்.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    அடிப்படை படம் மட்டும் குரங்கை கொண்டிருக்கும், ஆனால் இறுதி படம் குரங்கின் மீது தொப்பி இருக்கும்.
    
- **மாறுபாடுகளை உருவாக்கவும்**. உங்கள் அருகிலுள்ள படத்தை எடுத்து அதன் மாறுபாடுகள் உருவாக்கப்பட வேண்டும் என்று கேட்கலாம். ஒரு மாறுபாட்டை உருவாக்க, படம் ஒரு எழுத்து வழிகாட்டி மற்றும் கீழ்காணும் குறியீட்டை வழங்கவும்:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
