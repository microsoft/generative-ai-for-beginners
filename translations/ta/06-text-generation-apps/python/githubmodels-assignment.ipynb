# உரை உருவாக்கும் செயலிகள் கட்டமைக்கவும்

இந்த பாடத்திட்டத்தில் இதுவரை பார்த்தபடி, முன்மொழிவுகள் போன்ற அடிப்படை கருத்துகள் மற்றும் "prompt engineering" என்று ஒரு முழு பகுதி உள்ளத. நீங்கள் தொடர்பு கொண்ட பல கருவிகள், உதாரணமாக ChatGPT, Office 365, Microsoft Power Platform மற்றும் பிறவைப் போலவே, ஒரு காரியத்தைச் செய்வதற்கு முன்மொழிவுகளைப் பயன்படுத்த உங்களை ஆதரிக்கின்றன.

ஒரு செயலியில் இப்படிப் போன்ற அனுபவத்தைச் சேர்க்க, முன்மொழிவுகள், முடிவுகள் போன்ற கருத்துக்களை புரிந்து கொண்டு, செயல்பட ஒரு நூலகத்தைத் தேர்ந்தெடுக்க வேண்டும். இதையே நீங்கள் இந்த அத்தியாயத்தில் கற்றுக்கொள்ளப்போகின்றீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில் நீங்கள்:

- openai நூலகத்தையும் அதன் அடிப்படை கருத்துக்களையும் கற்றுக்கொள்ளலாம்.
- openai பயன்படுத்தி உரை உருவாக்கும் செயலி கட்டமைக்கலாம்.
- உரை உருவாக்கும் செயலியை கட்டமைக்க முன்மொழிவு, வெப்பநிலை, டோக்கன்கள் போன்ற கருத்துக்களை எவ்வாறு பயன்படுத்துவது என்பதை புரிந்துகொள்ளலாம்.

## கற்றல் குறிக்கோள்கள்

இந்த பாடத்தின் இறுதியில், நீங்கள்:

- உரை உருவாக்கும் செயலி என்ன என்பதை விளக்க முடியும்.
- openai பயன்படுத்தி உரை உருவாக்கும் செயலி கட்டமைக்க முடியும்.
- உங்கள் செயலியில் அதிகம் அல்லது குறைவாக டோக்கன்களை பயன்படுத்தவும், வெப்பநிலையை மாற்றி வெவ்வேறு வெளியீட்டை பெற்றுக் கொள்ளவும் அமைக்க முடியும்.

## உரை உருவாக்கும் செயலி என்றால் என்ன?

சாதாரணமாக நீங்கள் செயலி கட்டும்போது அதற்கு பின்வருவன போன்ற ஒரு இடைமுகம் இருக்கும்:

- கட்டளை-அடிப்படையாக. கட்டளை அடிப்படையிலான செயலிகள் என்பது நீங்கள் ஒரு கட்டளையை தட்டச்சு செய்து அது ஒரு செயலியைச் செய்யும் செயலிகள். உதாரணமாக, `git` ஒரு கட்டளை அடிப்படையிலான செயலி.
- பயனர் இடைமுகம் (UI). சில செயலிகளில் காட்சி பயனர் இடைமுகங்கள் (GUIs) உள்ளன, அதில் நீங்கள் பொத்தான்களை கிளிக் செய்யலாம், உரையை உள்ளீடு செய்யலாம், விருப்பங்களை தேர்ந்தெடுக்கலாம் மற்றும் மேலும் பல.

### கட்டளை மற்றும் UI செயலிகள் வரம்பற்றவை அல்ல

கட்டளை அடிப்படையிலான செயலி ஒன்றுடன் ஒப்பிடுங்கள்:

- **வரம்பற்றது**. நீங்கள் எந்தவொரு கட்டளையையும் தட்டச்சு செய்ய முடியாது, செயலி ஆதரிக்கும் கட்டளைகள் மட்டுமே.
- **மொழி_SPECIFIC**. சில செயலிகள் பல மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பாக ஒரு குறிப்பிட்ட மொழிக்காக செயலி கட்டப்பட்டுள்ளது, மேலும் நீங்கள் மேலும் மொழி ஆதரவைச் சேர்க்கலாம்.

### உரை உருவாக்கும் செயலிகளின் பலன்கள்

உரை உருவாக்கும் செயலி எப்படி வித்தியாசமாக இருக்கும்?

உரை உருவாக்கும் செயலியில், உங்களிடம் அதிகமான சுதந்திரம் உள்ளது, கட்டளைகளின் தொகுப்பிற்கோ அல்லது ஒரு குறிப்பிட்ட உள்ளீட்டு மொழிக்கோ நீங்கள் வரம்பற்றவர். பதிலுக்கு, இயற்கை மொழியைப் பயன்படுத்தி செயலியுடன் தொடர்பு கொள்ளலாம். மேலும் ஒரு நன்மை என்னவென்றால், நீங்கள் ஏற்கனவே பெருமளவு தகவலை அடிப்படையாக்கொண்ட ஒரு தரவு மூலதனத்துடன் தொடர்பு கொள்கிறீர்கள், ஆனால் பாரம்பரிய செயலி தரவுத்தளத்தில் என்ன உள்ளது என்பதிலேயே வரம்பமைக்கப்பட்டிருக்கலாம்.

### நான் உரை உருவாக்கும் செயலிகளுடன் என்ன செய்ய முடியும்?

பலவகைகளில் கட்டமைக்க முடியும். உதாரணமாக:

- **ஒரு chatbot**. உங்கள் நிறுவனம் மற்றும் அதன் விளைவுகளுக்கான கேள்விகளுக்கு பதிலளிக்கும் chatbot ஒரு நல்ல ஏற்பாடு ஆக இருக்கலாம்.
- **சேவையாளர்**. LLMகள் உரையை சுருக்குதல், உரையிலிருந்து தகவல் பெறுதல், ரெஸ்யூமே போன்ற உரைகளை உருவாக்குதல் போன்றவற்றுக்கு சிறந்தவை.
- **குறியீடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரியின்படி, குறியீடு எழுத உதவும் குறியீடு உதவியாளரை உருவாக்க முடியும். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளைப் பயன்படுத்தி குறியீடு எழுத உதவ முடியும்.

## நான் எப்படி தொடங்கலாம்?

LLM உடன் ஒருங்கிணைக்க விரும்பினால் நீங்கள் பொதுவாக பின்வரும் இரண்டு அணுகுமுறை வழிமுறைகள் வேண்டும்:

- ஒரு API பயன்படுத்தவும். இங்கே நீங்கள் உங்கள் முன்மொழிவுடன் இணைய வேண்டிய கோரிக்கைகளை உருவாக்கி உருவாக்கப்பட்ட உரையை பெறுகின்றீர்கள்.
- ஒரு நூலகம் பயன்படுத்தவும். நூலகங்கள் API அழைப்புகளை தொகுத்து எளிதாக பயன்படுத்த உதவுகின்றன.

### 1. Codespaces

## பயிற்சி - ஒரு சமையல் செய்முறை உருவாக்கி விடு

இப்போது நாமும் ஒரு காட்சி நிகழ்த்திவிட்டதால், அந்த காட்சிக்கு பொருந்தும் குறியீட்டை எழுதுவோம். இதற்கு பின்வரும் படிகளை பின்பற்றவும்:

1. உள்ளேற்றப்பட்ட கோப்பை துவக்கமாக பயன்படுத்துக
1. ஒரு `prompt` மாறியை உருவாக்கி கீழ்காணும் மாதிரியில் எடுத்துக்காட்டைக் குறியீட்டில் மாற்றுக:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

சிறந்தது, எப்படி நம்மால் விஷயங்களை மேம்படுத்தலாம் என்று பார்க்கலாம். விஷயங்களை மேம்படுத்த, குறியீடு நெகிழ்வுடையதாக இருக்க வேண்டும் என்பதைக் கண்டிப்பாக பார்க்க வேண்டும், ஆகவே பொருட்கள் மற்றும் சமையல் குறிப்புகளின் எண்ணிக்கை மேம்படுத்தபடக்கூடும் மற்றும் மாற்றப்படக்கூடும்.


1. கீழ்காணும் முறையில் கோ드를 மாற்றுவோம்:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
