# உரை உருவாக்கும் பயன்பாடுகளை உருவாக்குதல்

இக்குறிப்பெருக்கின் மூலம் நீங்கள் இதுவரை பார்த்திருப்பது போல, ப்ராம்ப்ட்கள் போன்ற முக்கிய கருத்துகள் மற்றும் "ப்ராம்ப்ட் இன்ஜினியரிங்" என்ற முழுமையான ஒரு துறை உள்ளது. ChatGPT, Office 365, Microsoft Power Platform மற்றும் இன்னும் பல தொடர்பு கொள்ளக்கூடிய கருவிகள், ஐதம்களை நிறைவேற்ற ப்ராம்ப்ட்களைப் பயன்படுத்தி உங்களுக்கு ஆதரவு அளிக்கின்றன.

இவ்வாறு ஒரு அனுபவத்தை ஒரு பயன்பாட்டில் சேர்க்க, ப்ராம்ப்ட், நிறைவு போன்ற கருத்துக்களைப் புரிந்து கொண்டு, பயன்படுத்த ஒரு நூலகத்தைத் தேர்ந்தெடுக்க வேண்டும். இதே விஷயத்தை நீங்கள் இந்த அத்தியாயத்தில் கற்றுக்கொள்ளப்போகிறீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் முக்கியக் கருத்துக்களைப் பற்றி கற்றுக்கொள்வீர்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்குவீர்கள்.
- ப்ராம்ப்ட், வெப்பநிலை மற்றும் டோக்கன்கள் போன்ற கருத்துக்களை பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை எவ்வாறு உருவாக்குவது என்பதைப் புரிந்து கொள்வீர்கள்.

## கற்றல் இலக்குகள்

இந்த பாடத்தின் முடிவில், நீங்கள்:

- ஒரு உரை உருவாக்கும் பயன்பாடு என்ன என்பதையாவது விளக்கமுடியும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்கலாம்.
- உங்கள் பயன்பாட்டை, குறைவான அல்லது அதிகமான டோக்கன்களைப் பயன்படுத்தவும், வெப்பநிலையையும் மாற்றவும், அப்படி மாறுபட்ட வெளியீட்டு பெற தனிப்பயன்படுத்தலாம்.

## உரை உருவாக்கும் பயன்பாடு என்பது என்ன?

பொதுவாக, நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும் போது அதை கீழ்கண்ட இன்டர்ஃபேஸ் வடிவத்திலோடு இருக்கும்:

- கட்டளை அடிப்படையிலானது. கட்டளை வரி பயன்பாடுகள் அதாவது நீங்கள் ஒரு கட்டளை இடும்போது அது ஒரு பணியைச் செய்யும். உதாரணமாக, `git` என்பது கட்டளை அடிப்படையிலான பயன்பாடு.
- பயனர் இடைமுகம் (UI). சில பயன்பாடுகள் குறும்படங்களுடன் GUI கொண்டிருக்குமே, இதில் நீங்கள் பொத்தான்களை கிளிக் செய்யலாம், உரையை உள்ளிடலாம், விருப்பங்களைத் தேர்ந்தெடுக்கலாம் மற்றும் இன்னும் பல.

### கட்டளை மற்றும் UI பயன்பாடுகள் வரம்பு குறைவு

ஒரு கட்டளை அடிப்படையிலான பயன்பாட்டுடன் ஒப்பிடுகையில், நீங்கள் கட்டளையைத் தட்டச்சு செய்வீர்கள்:

- **இது வரம்பு கொண்டது**. நீங்கள் எந்த கட்டளையையும் இட முடியாது, பயன்பாடு ஆதரிக்கும் கட்டளைகள் மட்டுமே இட முடியும்.
- **மொழி குறிப்பிட்டது**. சில பயன்பாடுகள் நிறைய மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பான முறையில் அது ஒரு குறிப்பிட்ட மொழிக்காக உருவாக்கப்பட்டுள்ளது, கூடுதல் மொழி ஆதரவைச் சேர்க்கலாம்.

### உரை உருவாக்கும் பயன்பாடுகளின் நன்மைகள்

எனவே, உரை உருவாக்கும் பயன்பாடு எப்படி வேறுபடிறது?

ஒரு உரை உருவாக்கும் பயன்பாட்டில், நீங்கள் அதிகமான தடைகள் இல்லாமல் செயல்பட முடியும், கட்டளைகள் சிலருக்கு மட்டும் அல்லாமல் இயற்கை மொழியைக் கொண்டு பயன்பாடுடன் தொடர்பு கொள்ளலாம். மேலும், நீங்கள் ஏற்கனவே தகவல்களின் பரந்த தொகுப்பறியில் பயிற்சி பெற்ற தரவுத் தளத்துடன் தொடர்பு கொண்டிருக்கிறீர்கள், இது பாரம்பரிய பயன்பாடுகளுக்கு எல்லையிடப்பட்ட தரவுத்தளத்திலிருந்து வேறுபட்டது.

### உரை உருவாக்கும் பயன்பாட்டுடன் நான் என்ன செய்ய முடியும்?

பல விஷயங்களை உருவாக்க முடியும். உதாரணமாக:

- **ஒரு chattebot**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் பற்றி கேள்விகளுக்கு பதிலளிக்கும் ஒரு chattebot நல்ல தேர்வாக இருக்கலாம்.
- **உதவியாளர்**. LLMக்கள் உரைகளை சுருக்குவது, உரையில் இருந்து தகவல்கள் பெறுவது, ரெசுமே போன்ற உரைகளை உருவாக்குவது போன்றவற்றில் சிறந்தவை.
- **குறியீடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரியைப் பொறுத்து, குறியீடு எழுத உதவும் உதவியாளரை உருவாக்கலாம். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளை குறியீடு எழுத ஓர் உதவியாக பயன்படுத்தலாம்.

## நான் எப்படி துவங்குவது?

பொதுவாக, LLM உடன் ஒருங்கிணைப்பதை கண்டுபிடிக்க வேண்டும், இது பொதுவாக இரண்டு வழிகளைக் கொண்டுள்ளது:

- API ஐ பயன்படுத்துக. இங்கு நீங்கள் உங்களது ப்ராம்ப்டுடன் வலை கோரிக்கைகளை உருவாக்கி உருவாக்கப்பட்ட உரையை பெறுகிறீர்கள்.
- ஒரு நூலகத்தை பயன்படுத்துக. நூலகங்கள் API அழைப்புக்களைச் சுருக்கி அவற்றை எளிதாக பயன்படுத்த உதவுகின்றன.

## நூலகங்கள்/SDKகள்

LLMக்களுடன் வேலை செய்வதற்கான சில பரிச்சயமான நூலகங்கள் உள்ளன:

- **openai**, இந்த நூலகம் உங்கள் மாதிரிக்கு எளிதில் இணைக்கவும் ப்ராம்ப்ட்களை அனுப்பவும் உதவும்.

பிறகு மேலாண்மை நிலை நூலகங்கள் உள்ளன, உதாரணமாக:

- **Langchain**. Langchain பரிச்சயமானது மற்றும் Python ஆதரிக்கிறது.
- **Semantic Kernel**. Semantic Kernel என்பது Microsoft வழங்கும் நூலகம், C#, Python, மற்றும் Java மொழிகளை ஆதரிக்கிறது.

## பயிற்சி - ஒரு சமையல் செய்முறை உருவாக்கி எழுது

இப்போது நாங்கள் ஒரு காட்சி நிகழ்ச்சியை விளையாடிவிட்டோம், காட்டப்பட்ட காட்சிக்கு பொருந்தும் குறியீட்டை எழுதுவோம். அவ்வாறு செய்வதற்கு, பின்வரும் படிகளை பின்பற்றவும்:

1. உள்ளடக்க கோப்பைப் பாவித்து தொடங்கவும்
1. `prompt` என்ற 변수 உருவாக்கி கீழ்க்கண்ட மாதிரி குறியீட்டை மாற்றவும்:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. கீழ்காணும் முறையில் கோட்டை மாற்றுவோம்:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
