# உரை உருவாக்கும் பயன்பாடுகளை உருவாக்கவும்

இந்த பாடத்திட்டத்தில் இதுவரை நீங்கள் பார்த்துள்ளீர்கள், அங்கே ప్రాంప్ போன்ற முக்கிய கருத்துக்கள் மற்றும் "prompt engineering" எனப்படும் ஒரு முழு துறை இதன் இணைப்பு உள்ளது. ChatGPT, Office 365, Microsoft Power Platform மற்றும் பல உபகரணங்கள் போன்றவை, உரைகளை பயன்படுத்தி ஏதேனும் ஒன்றை சாதிக்க உதவுகின்றன.

உங்கள் பயன்பாட்டில் இப்படியான அனுபவத்தை சேர்க்க, prompts, completions போன்ற கருத்துக்களைப் புரிந்துகொள்ள வேண்டும் மற்றும் பணியாற்ற ஒரு நூலகத்தை தேர்ந்தெடுக்க வேண்டும். இதே உண்மையை இந்த அத்தியாயத்தில் நீங்கள் கற்றுக்கொள்ளப்போகிறீர்கள்.

## அறிமுகம்

இந்த அத்தியாயத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் முக்கிய கருத்துக்களைப் பற்றி கற்றுக்கொள்ளவும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்கவும்.
- prompt, temperature, tokens போன்ற கருத்துக்களை பயன்படுத்தி உரை உருவாக்கும் பயன்பாட்டை எப்படி உருவாக்குவது என்பதை புரிந்துகொள்ளவும்.

## கற்றல் குறிக்கோள்கள்

இந்த பாடம் முடிவில், நீங்கள் முடியும்:

- ஒரு உரை உருவாக்கும் பயன்பாடு என்பது என்ன என்பதை விளக்கவும்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்கவும்.
- உங்கள் பயன்பாட்டை அதிகமான அல்லது குறைந்த tokens பயன்படுத்தவும் மற்றும் temperature மாற்றவும், மாற்றுமா உள்ளெடுப்புக்காக.

## உரை உருவாக்கும் பயன்பாடு என்றால் என்ன?

சாதாரணமாக நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும் போது அதை கீழ்க்காணும் வகையான இடைமுகம் உள்ளது:

- கட்டளையைக் கொண்டு இயங்கும். Console பயன்பாடுகள் பொதுவாக நீங்கள் கட்டளையை type செய்தால் அது அந்த பணியை செய்து முடிக்கும். உதாரணமாக, `git` என்பது கட்டளைக் கொண்டு இயங்கும் பயன்பாடு.
- பயனர் இடைமுகம் (UI). சில பயன்பாடுகளில் graphical user interfaces (GUIs) உள்ளது, நீங்கள் பட்டன்களை அழுத்தி, உரையை உள்ளிடி, தேர்வுகளைத் தெரிவு செய்து பயன்படுத்துகிறீர்கள்.

### Console மற்றும் UI பயன்பாடுகள் வரையறுக்கப்பட்டவை

கட்டளையாக type செய்யும் app உடன் ஒப்பிடுக:

- **அவை வரையறுக்கப்பட்டவை**. எல்லா கட்டளையும் type செய்ய முடியாது, பயன்பாடு அனுமதிக்கும் கட்டளைகள் மட்டுமே பயன்படுத்த முடியும்.
- **மொழி சார்ந்தவை**. சில பயன்பாடுகள் பல மொழிகளை ஆதரிக்கின்றன, ஆனால் இயல்பாக அந்த பயன்பாடு குறிப்பிட்ட ஒரு மொழிக்கு உருவாக்கப்பட்டுள்ளது, கூடுதலாக மொழி ஆதரவு சேர்க்க முடியும் என்றாலும்.

### உரை உருவாக்கும் பயன்பாடுகளின் நன்மைகள்

ஒரு உரை உருவாக்கும் பயன்பாடு எப்படி வேறுபடுகிறது?

ஒரு உரை உருவாக்கும் பயன்பாட்டில், உங்களுக்கு அதிக அனுமதி இருக்கிறது, நீங்கள் கட்டளைகளின் தொகுப்புக்கு அல்லது ஒரு குறிப்பிட்ட உள்ளீட்டு மொழிக்கு வரையறுக்கப்படவில்லை. பதிலாக, நீங்கள் இயல்பான மொழி பயன்படுத்தி பயன்பாட்டுடன் தொடர்பு கொள்ள முடியும். மற்றொரு நன்மை, நீங்கள் ஏற்கனவே பெரிய தகவல் தொகுப்பில் பயிற்சி பெற்ற தரவுத்தளத்துடன் தொடர்பு கொள்கிறீர்கள், ஆனால் பாரம்பரிய பயன்பாடுகள் தரவுத்தளத்தில் உள்ளவற்றில் மட்டுமே வரையறுக்கபடுகின்றன.

### நான் உரை உருவாக்கும் பயன்பாட்டுடன் என்ன கட்டமைக்க முடியும்?

நீங்கள் பல விஷயங்களை உருவாக்க முடியும். உதாரணமாக:

- **ஒரு chatbot**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் பற்றி கேள்விகளுக்கு பதில் அளிக்கும் chatbot நல்ல பொருத்தமாக இருக்கும்.
- **உதவியாளர்**. LLMகள் உரையை சுருக்குதல், உள்ளடக்கத்தைப் புரிதல், உயிரணுக்கான உரை உருவாக்குதல் போன்றவற்றில் சிறந்தவை.
- **கோடு உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரிக்கு ஏற்ப, நீங்கள் ஒரு கோடு உதவியாளரை உருவாக்கலாம். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற பொருட்களைக் கொண்டு நீங்கள் கோடு எழுத உதவி பெறலாம்.

## நான் எப்படி துவங்கலாம்?

நீங்கள் LLM உடன் பொருந்த வரும் வகையில் வழியை கண்டுபிடிக்க வேண்டும், இது பொதுவாக கீழ்க்காணும் இரண்டு அணுகுமுறைகளைக் கொண்டுள்ளது:

- API பயன்படுத்தவும். இங்கு நீங்கள் உங்கள் prompt உடன் வலை கோரிக்கைகளை உருவாக்கி உருவாக்கப்பட்ட உரையைப் பெறுகிறீர்கள்.
- நூலகம் பயன்படுத்தவும். நூலகங்கள் API அழைப்புகளை மூடியவையாகச் செய்து அவற்றை சுலபமாகக் கையாள உதவுகின்றன.

## நூலகங்கள்/SDKகள்

LLMகளுடன் பணியாற்ற சில நன்றாகப் பரிச்சயமான நூலகங்கள் உள்ளன:

- **openai**, இந்த நூலகம் உங்கள் மாதிரியை எளிதாக இணைக்கவும், prompts அனுப்பவும் உதவுகிறது.

பின்னர் மேல்நிலை செயல்பாடுகளை வழங்கும் நூலகங்கள் உள்ளன:

- **Langchain**. Langchain பிரபலமாக உள்ளது மற்றும் Python ஐ ஆதரிக்கிறது.
- **Semantic Kernel**. Semantic Kernel என்பது Microsoft வழங்கும் நூலகம், இது C#, Python மற்றும் Java மொழிகளைக் கைகொண்டுள்ளது.

## openai பயன்படுத்தி முதல் பயன்பாடு

நாம் எவ்வாறு முதல் பயன்பாட்டை உருவாக்குவது, எத்தனை நூலகங்கள் தேவையென்று பார்த்துவிடலாம்.

### openai ஐ நிறுவவும்

  > [!NOTE] Codespaces அல்லது Devcontainer இல் இந்த நோட்புக் ஓட்டினால் இந்த படி தேவையில்லை


OpenAI அல்லது Azure OpenAI உடன் தொடர்பு கொள்ள பல நூலகங்கள் உள்ளன. C#, Python, JavaScript, Java போன்ற பல வேலை மொழிகளையும் பயன்படுத்த முடியும்.  
நாம் `openai` Python நூலகம் பயன்படுத்த தேர்ந்தெடுத்துள்ளோம், எனவே `pip` கொண்டு அதை நிறுவப்போகிறோம்.

```bash
pip install openai
```

நீங்கள் Codespaces அல்லது Dev Container இல் இந்த நோட்புக் ஓட்டவில்லையெனில், உங்கள் இயந்திரத்தில் [Python](https://www.python.org/) ஐ நிறுவ வேண்டும்.

### ஒரு ரிசோர்ஸ் உருவாக்கி API விசையை கண்டறியவும்

நீங்கள் இப்போதतकா செய்யவில்லை என்றால், கீழ்க்காணும் படிகளைச் செய்ய வேண்டும்:

- OpenAI இல் ஒரு கணக்கை உருவாக்கவும் <https://platform.openai.com/signup>.
- இப்போது, உங்கள் API விசைகளைப் பெறவும் <https://platform.openai.com/api-keys>. 

>[!NOTE]
> உங்கள் API விசையை உங்கள் கோட்டிலிருந்து பிரித்துச் சேர்க்கValueல் உகந்தது. நீங்கள் சூழல் மாறிகள் கொண்டு இதைக் செய்யலாம்.
> - .env கோப்பில் `OPENAI_KEY` என்ற சூழல் மாறி உங்கள் API விசைக்கு அமைக்கவும். நீங்கள் ஏற்கனவே இந்த பாடக்கிருமையின் முந்தைய பயிற்சிகளை முடித்திருந்தால், நீங்கள் தயார்.
> - API விசை ஒருமுறை மட்டுமே பெறக்கூடியது என்பதை கவனியுங்கள். இதைப் பிழையானவாறு அச்சிடுகிறீர்களானால், அதை நீக்கி புதிய விசையை உருவாக்கவும்.


### OpenAI அமைப்பை அமைக்கவும்

நீங்கள் OpenAI பயன்படுத்தினால், நீங்கள் அமைப்பை பின்வருமாறு செய்யலாம்:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

மேலே நாம் என்ன செய்வதெனில்:

- `api_key`, இது OpenAI டாஷ்போர்டில் உள்ள உங்கள் API விசை.
- `deployment`, இது உங்கள் GPT பதிப்பு.

> [!NOTE]
> `os.environ` என்பது சூழல் மாறிகளை வாசிக்கும் ஒரு செயல்பாடு. இதை `OPENAI_API_KEY` போன்ற சூழல் மாறிகளை வாசிக்க பயன்படுத்தலாம்.

## உரை உருவாக்குதல்

உரை உருவாக்க `responses.create` முறை பயன்படுத்தப்படும். உதாரணம்:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

மேலே உள்ள குறியீட்டில், நாம் ஒரு பதில் பொருளை உருவாக்கி, எந்த மாதிரியைப் பயன்படுத்த விரும்புகிறோமோ மற்றும் prompt ஐ அனுப்புகிறோம். பிறகு உருவாக்கப்பட்ட உரையை அச்சிடுகிறோம்.

### உரையாடல் பதில்கள்

இதுவரை, நீங்கள் ஒரு பிராம்ப்டிலிருந்து உரையை உருவாக்குங்கல் எப்படி என்பதைப் பார்த்தீர்கள். Responses API chatbot க்கான சிறந்தது - நீங்கள் `input` ஆக பல செய்திகள் பட்டியலை வழங்கி உரையாடலை கட்டமைக்கலாம். இதற்கான உதாரணம்:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

இதற்கான மேலதிக தகவல்கள் வரும் அத்தியாயத்தில் சரிபார்க்கப்படும்.
