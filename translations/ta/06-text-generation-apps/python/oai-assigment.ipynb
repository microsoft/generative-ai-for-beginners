# உரை உருவாக்கும் செயலிகளைக் கட்டமைக்கவும்

இக்கரிக்குலத்தில் இதுவரை நீங்கள் பார்த்துள்ளீர்கள் போல, உத்தேசங்கள் போன்ற அடிப்படைக் கருத்துகள் மற்றும் "உத்தேசக் கலைபடுத்தல்" எனப்படும் ஒரு முழுத் துறையும் உள்ளது. ChatGPT, Office 365, Microsoft Power Platform மற்றும் பலவற்றைப் போன்ற பல கருவிகள், உத்தேசங்களைப் பயன்படுத்தி எதையாவது சாதிக்க உதவுகின்றன.

ஒரு செயலியில் இப்படியான அனுபவத்தை சேர்க்க, உத்தேசங்கள், முழுமைகள் போன்ற கருத்துக்களைப் புரிந்துகொண்டு, பணியாற்ற ஒரு நூலகத்தை தேர்ந்தெடுக்க வேண்டும். இதுதான் நீங்கள் இந்த அதிகாரத்தில் கற்றுக்கொள்ளப்போகிறீர்கள்.

## அறிமுகம்

இந்த அதிகாரத்தில் நீங்கள்:

- openai நூலகம் மற்றும் அதன் அடிப்படைக் கருத்துக்களைப் பற்றி கற்றுக்கொள்ளுங்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் செயலியை கட்டமைக்கவும்.
- உத்தேசம், வெப்பநிலை மற்றும் டோக்கன்கள் போன்ற கருத்துக்களைச் பயன்படுத்தி எப்படி உரை உருவாக்கும் செயலியை உருவாக்குவது என்பதைப் புரிந்துகொள்ளுங்கள்.

## கற்றல் இலக்குகள்

இந்த பாடத் துறையின் முடிவில், நீங்கள் செய்ய முடியும்:

- ஒரு உரை உருவாக்கும் செயலி என்ன என்பது விளக்குங்கள்.
- openai பயன்படுத்தி ஒரு உரை உருவாக்கும் செயலியை கட்டமைக்கவும்.
- உங்கள் செயலியில் அதிக அல்லது குறைந்த டோக்கன்களை பயன்படுத்தும் வகையில் கட்டமைக்கவும், வெப்பநிலையை மாற்றவும், மாறுபட்ட விளைவிற்கு.

## உரை உருவாக்கும் செயலி என்றால் என்ன?

சாதாரணமாக, ஒரு செயலி கட்டும்போது அதற்கு கீழ்காணும் வகையான இடைமுகம் இருக்கும்:

- கட்டளை அடிப்படையிலானது. கான்சோல் செயலிகள் என்றால் நீங்கள் ஒரு கட்டளையை تایப் செய்து ஒரு பணி செய்யும் செயலிகள். உதாரணமாக, `git` என்பது கட்டளை அடிப்படையிலான செயலி.
- பயனர் இடைமுகம் (UI). சில செயலிகளுக்கு கிளிக் பட்டு, உரை உள்ளிடவும், விருப்பங்களை தேர்ந்தெடுக்கவும் மற்றும் பலவற்றுக்கான கிராபிக்கல் பயனர் இடைமுகம் (GUI) கருவிகள் உண்டு.

### கான்சோல் மற்றும் UI செயலிகளுக்கு வரம்பு உள்ளது

கட்டளை تایப் செய்யும் செயலியை ஒப்பிடுங்கள்: 

- **வரம்பளவு** உள்ளது. நீங்கள் எந்த கட்டளையையும் تایப் செய்ய முடியாது, செயலி ஆதரிக்கும் கட்டளைகள் மட்டும்.
- **மொழி குறிப்பிட்டது**. சில செயலிகள் பல மொழிகளுக்கு ஆதரவு அளிக்கின்றன, ஆனால் இயல்பாக செயலி ஒரு குறிப்பிட்ட மொழிக்கானதாக கட்டமைக்கப்பட்டுள்ளது, கூடுதலாக மொழி ஆதரவு சேர்க்கப்பட்டாலும். 

### உரை உருவாக்கும் செயலிகளின் நன்மைகள்

ஆகவே, ஒரு உரை உருவாக்கும் செயலி எப்படி வேறுபடுகிறது?

ஒரு உரை உருவாக்கும் செயலியில், நீங்கள் அதிக வளைவு கொண்டீர்கள், ஒரு கட்டளை தொகுப்பிற்கு அல்லது குறிப்பிட்ட உள்ளீட்டு மொழிக்கு மட்டுமல்ல. பதிலாக, இயற்கை மொழியைப் பயன்படுத்தி செயலியுடன் தொடர்பு கொள்ளலாம். மேலும், ஏற்கனவே ஒரு பெரிய தகவல் தொகுப்பு பயிற்சி பெற்ற தரவுத்தளத்துடன் நீங்கள் செயல்படுகிறீர்கள், ஒரு பாரம்பரிய செயலி ஒரு தரவுத்தளத்தில் உள்ளதிற்கே வரம்பானதாக இருக்கலாம்.

### நான் உரை உருவாக்கும் செயலியுடன் என்ன செய்ய முடியும்?

நீங்கள் பலவற்றை உருவாக்கலாம். உதாரணமாக:

- **ஒரு குறும்படம்**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் குறித்து கேள்விகளுக்கு பதிலளிக்கும் குறும்படம் நல்ல பொருத்தமாக இருக்கும்.
- **உதவியாளர்**. LLMக்கள் உரையை சுருக்குவது, உரையிலிருந்து洞察ங்களை பெறுவது, ரெஜூமே போன்ற உரைகளை உருவாக்குவது போன்ற விஷயங்களில் சிறந்தவை.
- **குறியீடு உதவியாளர்**. நீங்கள் எவ்வளவு மொழி மாதிரியை பயன்படுத்துகிறீர்களோ, அதன்படி குறியீடு எழுத உதவும் குறியீடு உதவியாளரை உருவாக்கலாம். உதாரணமாக, GitHub Copilot மற்றும் ChatGPT போன்ற தயாரிப்புகளை பயன்படுத்தி குறியீடு எழுத உதவிக்கலாம்.

## எப்படி தொடங்குவது?

நல்லது, LLM உடன் ஒருங்கிணைக்க இரண்டு முறைகள் இருக்கிறது:

- API ஐப் பயன்படுத்தவும். இங்கு நீங்கள் உங்கள் உத்தேசத்துடன் வலை கோரிக்கைகளை உருவாக்கி உருவாக்கப்பட்ட உரையை பெறுகிறீர்கள்.
- நூலகம் பயன்படுத்தவும். நூலகங்கள் API அழைப்புகளை முழுமையாக கட்டுப்படுத்து முறையில் இருந்து பெற்று பயன்படுத்த எளிதாக்குகின்றன.

## நூலகங்கள்/SDKகள்

LLMக்களுடன் பணியாற்ற பிரபலமான சில நூலகங்கள் உள்ளன:

- **openai**, இந்த நூலகம் உங்கள் மாதிரியை இணைக்கவும் உத்தேசங்களை அனுப்பவும் எளிதாக்குகிறது.

பிறகு மேல்நிலை இயங்கும் நூலகங்கள் இருக்கும்:

- **Langchain**. Langchain பிரபலமாக மற்றும் Python ஐ ஆதரிக்கிறது.
- **Semantic Kernel**. Semantic Kernel என்பது Microsoft உருவாக்கிய நூலகம், C#, Python, மற்றும் Java ஆகிய மொழிகளை ஆதரிக்கிறது.

## openai பயன்படுத்தி முதல் செயலி

நம்முடைய முதல் செயலியை எப்படி உருவாக்கலாம், எந்த நூலகங்கள் தேவை, எவ்வளவு தேவை போன்றவை பார்க்கலாம்.

OpenAI அல்லது Azure OpenAI உடன் தொடர்பு கொள்ள பல நூலகங்கள் உள்ளன. C#, Python, JavaScript, Java மற்றும் மேலும் பல நிரலாக்க மொழிகள் பயன்படுத்தக்கூடியவை.  
நாம் `openai` Python நூலகத்தை தேர்ந்தெடுத்தோம், எனவே `pip` கொண்டு நிறுவுவோம்.

```bash
pip install openai
```

நீங்கள் Codespaces அல்லது Dev Containerல் இந்த நோட்புக்கை இயக்கவில்லை என்றால், உங்கள் கணினியில் [Python](https://www.python.org/)யை நிறுவவும்.

### ஒரு வளத்தை உருவாக்கி உங்கள் API விசையை கண்டறியவும்

நீங்கள் இதுவரை செய்யவில்லை என்றால், கீழ்க்கண்ட படிகளைச் செய்ய வேண்டும்:

- OpenAI இல் கணக்கு உருவாக்கவும் <https://platform.openai.com/signup>.
- இப்போது, உங்கள் API விசைகளைப் பெறுக <https://platform.openai.com/api-keys>. 

>[!NOTE]
> உங்கள் API விசையை உங்கள் குறியீட்டிலிருந்து பிரிக்க வேண்டியது நல்லது. இதற்கு சுற்றுச்சூழல் மாறிலிகள் பயன்படுத்தலாம்.
> - `.env` கோப்பில் `OPENAI_KEY` சுற்றுச்சூழல் மாறிலியை உங்கள் API விசையாக அமைக்கவும். நீங்கள் ஏற்கனவே இந்த பாடநெறியின் பழைய பயிற்சிகளை முடிச்சில் இருக்கிறீர்கள் என்றால், எல்லாம் தயார்.
> - API விசை ஒருமுறை மட்டுமே அணுககூடியது என்பதைக் கவனிக்கவும். சரியாக நகலெடுக்கப்பட்டிருக்கிறதா என உறுதிப்படுத்த வேண்டும். அது சரியாக வேலை செய்யவில்லையெனில், அதை நீக்கி, புதிய விசையை உருவாக்க பரிந்துரைக்கப்படுகிறது.


### OpenAI சூழலை அமைக்கவும்

நீங்கள் OpenAI பயன்படுத்தினால், சூழலை இப்படிக் கட்டமைக்கலாம்:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

மேலே நாம் பின்வருபவை அமைக்கிறோம்:

- `api_key`, இது OpenAI டாஷ்போர்டில் காணப்படும் உங்கள் API விசை.
- `deployment`, இது உங்கள் GPT பதிப்பு.

> [!NOTE]
> `os.environ` என்பது சுற்றுச்சூழல் மாறிலிகளை வாசிக்கும் ஒரு செயல்பாடு. நீங்கள் இதைப் பயன்படுத்தி `OPENAI_API_KEY` போன்றவற்றை வாசிக்கலாம்.

## உரையை உருவாக்கவும்

உரையைக் உருவாக்க `responses.create` முறையைப் பயன்படுத்த வேண்டும். உதாரணம்:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

மேலுள்ள குறியீட்டில், நாம் ஒரு பதில் பொருளை உருவாக்கி, பயன்படுத்த வேண்டிய மாதிரியும் உத்தேசத்தையும் வழங்குகிறோம். பின்னர் உருவாக்கப்பட்ட உரையை அச்சிடுகிறோம்.

### உரையாடல் பதில்கள்

இப்போது வரை, ஒரு உத்தேசத்திலிருந்து உரையை உருவாக்க எப்படி என்று பார்த்தோம். Responses API குறும்படங்களுக்கு (chatbots) கூட நல்லதாகும் - நீங்கள் உரையாடலை கட்டமைக்க பல செய்திகளைக் `input` ஆக வழங்கலாம். இது பயன்படுத்தும் உதாரணம்:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

இதைப் பயன்படுத்துவதற்கான மேலதிக விளக்கங்கள் வரும் அதிகாரத்தில் இருக்கும்.
