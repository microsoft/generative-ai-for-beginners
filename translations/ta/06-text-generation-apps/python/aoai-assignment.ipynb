# உரை உருவாக்கும் பயன்பாடுகளை உருவாக்குதல்

நீங்கள் இதுவரை இந்த பாடத்திட்டத்தில் பார்த்து வந்திருப்பது போல், முன் வினாக்கள் போன்ற அடிப்படைக் கருத்துகளும் "முன்வினை பொறியியல்" என்று முழு ஒரு துறை கூட உள்ளது. ChatGPT, Office 365, Microsoft Power Platform உள்ளிட்ட பல கருவிகள், உங்கள் முன் வினாக்களை பயன்படுத்தி ஏதாவது செய்ய உதவுகின்றன.

ஒரு பயன்பாட்டுக்கு இத்தகைய அனுபவத்தை சேர்க்க, நீங்கள் முன் வினாக்கள், முடிவுகள் போன்ற கருத்துக்களை புரிந்து கொண்டு, ஒரு நூலகத்தை தேர்ந்தெடுத்துப் பயன்படுத்த வேண்டும். இதுதான் இந்த அதிகாரத்தில் நீங்கள் கற்றுக்கொள்ளப்போகும் விஷயம்.

## அறிமுகம்

இந்த அதிகாரத்தில், நீங்கள்:

- openai நூலகம் மற்றும் அதன் அடிப்படைக் கருத்துக்களைப் பற்றி அறிந்து கொள்ளவிருப்பீர்கள்.
- openai-ஐப் பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்குவீர்கள்.
- முன் வினை, வெப்பநிலை மற்றும் டோக்கன்கள் போன்ற கருத்துகளை எப்படி பயன்படுத்தி உரை உருவாக்கும் பயன்பாட்டை உருவாக்கலாம் என்பதையும் புரிந்து கொள்வீர்கள்.

## கற்றல் குறிக்கோள்கள்

இந்த பாடம் முடிவில், நீங்கள்:

- ஒரு உரை உருவாக்கும் பயன்பாடு என்றால் என்ன என்று விளக்க முடியும்.
- openai-ஐப் பயன்படுத்தி ஒரு உரை உருவாக்கும் பயன்பாட்டை உருவாக்க முடியும்.
- மேலும் அல்லது குறைவான டோக்கன்களை பயன்படுத்தவும், வெப்பநிலையை மாற்றவும் உங்கள் பயன்பாட்டை அமைக்க முடியும், இதனால் வேறுபட்ட வெளியீடு பெறலாம்.

## உரை உருவாக்கும் பயன்பாடு என்றால் என்ன?

சாதாரணமாக நீங்கள் ஒரு பயன்பாட்டை உருவாக்கும் போது அதில் பின்வருமாறு சில வகை இடைமுகங்கள் இருக்கும்:

- கட்டளை-அடிப்படையிலானது. கான்சோல் பயன்பாடுகள் என்பது நீங்கள் ஒரு கட்டளை அலங்கரித்து செயல்படுத்தும் பயன்பாடுகள். உதாரணமாக, `git` என்பது கட்டளை-அடிப்படையிலான பயன்பாடு.
- பயனர் இடைமுகம் (UI). சில பயன்பாடுகள் குறும்பட பயனர் இடைமுகங்களை கொண்டிருக்கும், அங்கு நீங்கள் பொத்தான்களை கிளிக் செய்வீர்கள், உரையை உள்ளிடுவீர்கள், விருப்பங்களை தேர்ந்தெடுப்பீர்கள் மற்றும் பல.

### கான்சோல் மற்றும் UI பயன்பாடுகள் வரம்பாக உள்ளன

கட்டளை-அடிப்படையிலான பயன்பாட்டுக்குப் பதிலாக அதைப் ஒப்பிடுக:

- **இது வரம்புள்ளதானது**. நீங்கள் எவ்வித குடிச்சட்டத்தையும் தட்ட முடியாது, பயன்பாடு ஆதரவளிக்கும் கட்டளைகள் மட்டுமே.
- **மொழி-சந்தி**. சில பயன்பாடுகள் பல மொழிகளை ஆதரிக்கும், ஆனால் இயல்பாக பயன்பாடு ஒரு குறிப்பிட்ட மொழிக்காக கட்டமைக்கப்பட்டுள்ளது, கூடுதலான மொழி ஆதரவு சேர்க்கலாம் என்றாலும்.

### உரை உருவாக்கும் பயன்பாடுகளின் நன்மைகள்

எனவே, உரை உருவாக்கும் பயன்பாடு எப்படி வேறுபடுகிறது?

உரை உருவாக்கும் பயன்பாட்டில், உங்கள் வசதி அதிகமாக இருக்கும், நீங்கள் கட்டளைகள் அல்லது குறிப்பிட்ட உள்ளீட்டு மொழியில் வரம்படையப்பட மாட்டீர்கள். அதற்கு பதிலாக, இயல்பு மொழியை பயன்படுத்தி பயன்பாட்டுடன் தொடர்பு கொள்ளலாம். மேலும், நீங்கள் ஏற்கனவே பரந்த தொகுப்பான தகவல் முதன்மையாக பயிற்சி பெறப்பட்ட தரவு மூலத்துடன் தொடர்பு கொண்டுள்ளீர்கள், எனவே பாரம்பரிய பயன்பாடு தரவுத்தளத்தில் உள்ளதுதான் வரம்பாக இருக்கும்.

### உரை உருவாக்கும் பயன்பாட்டோடு என்ன செய்ய முடியும்?

நீங்கள் பலவகை செயல்களை உருவாக்க முடியும். உதாரணத்திற்கு:

- **ஒரு சாட்போட்**. உங்கள் நிறுவனம் மற்றும் அதன் தயாரிப்புகள் பற்றிய கேள்விகளுக்கு பதிலளிக்கும் சாட்போட் போப்பது நன்று.
- **உதவியாளர்**. LLM-க்கள் உரையை சுருக்க, உரையிலிருந்து தகவல்கள் பெற, பார்வையாளர்கள் போன்ற உரைகளை உருவாக்கும் போன்றவற்றில் சிறந்தவை.
- **கோட் உதவியாளர்**. நீங்கள் பயன்படுத்தும் மொழி மாதிரி அடிப்படையில், நீங்கள் கோட் எழுத உதவும் உதவியாளரை உருவாக்க முடியும். உதாரணமாக, GitHub Copilot போன்ற தயாரிப்புகளை மற்றும் ChatGPT-ஐக் கொண்டு கோட் எழுத உதவலாம்.

## நான் எப்படி துவங்குவது?

நீங்கள் LLM-ஐ இணைக்குவதற்கான வழிசொல்ல வேண்டும், பொதுவாக இரண்டு வழிமுறைகள் உள்ளன:

- API-ஐ பயன்படுத்துக. இங்கே நீங்கள் உங்கள் முன்வினை அனைத்தையும் கொண்டு வெப் கோரிக்கைகளை உருவாக்கி, உரை உருவாக்கப்படும் பதிலை பெறுவீர்கள்.
- நூலகத்தைப் பயன்படுத்துக. நூலகங்கள் API அழைப்புகளை மூடிவைத்து, அவற்றை பயன்படுத்த எளிதாக்குகின்றன.

LLM-களுடன் வேலை செய்ய சில பரவலான நூலகங்கள் உள்ளன:

## openai பயன்படுத்தி முதல் பயன்பாடு

நாம் எவ்வாறு முதல் பயன்பாட்டை உருவாக்குவது, எந்த நூலகங்கள் தேவை, எவ்வளவு தேவை என்பதைக் காணோம்.

### openai ஐ நிறுவுக

  > [!NOTE] இந்த படி Codespaces-ல் அல்லது Devcontainer-ல் இதை இயக்கவில்லை என்றால் அவசியம்.


OpenAI அல்லது Azure OpenAI உடன் தொடர்பு கொள்ள பல நூலகங்கள் உள்ளன. C#, Python, JavaScript, Java போன்ற பல கணினி மொழிகளை பயன்படுத்த முடியும்.  
நாங்கள் `openai` பைதான் நூலகத்தை பயன்படுத்தத் தீர்மானித்துள்ளோம், ஆகையால் அதை நிறுவ `pip`-ஐ பயன்படுத்துவோம்.

```bash
pip install openai
```

Codespaces அல்லது Dev Container-ல் இந்த நோட்புத்தகத்தை இயக்கவில்லையெனில், உங்கள் கணினியில் [Python](https://www.python.org/) ஐ நிறுவ வேண்டும்.

### ஒரு வளத்தை உருவாக்குக

நீங்கள் இதுவரை செய்யவில்லை என்றால், பின்வரும் படிகளை மேற்கொள்ள வேண்டும்:

- Azure-ல் ஒரு கணக்கை உருவாகவும் <https://azure.microsoft.com/free/>
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) இல் Azure OpenAI வளத்தை உருவாக்கவும். [வளத்தை உருவாக்குவது எப்படி](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) என்ற வழிகாட்டியைக் காணவும்.


### API கீவும் இறுகுகையும் கண்டறியவும்

இப்போது, `openai` நூலகத்திற்கு எந்த API கீவை பயன்படுத்த வேண்டும் என்று கூற வேண்டும். உங்கள் API கீவை கண்டுபிடிக்க, உங்கள் Azure OpenAI வளத்தின் "Keys and Endpoint" பகுதியை திறந்து "Key 1" மதிப்பை நகலெடுக்கவும்.

  ![Azure போர்டலில் Keys and Endpoint வள பக்கம்](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

இப்போது இந்த தகவலை நகலெடுத்துள்ளீர்கள், நூலகங்கள் அதை பயன்படுத்தச் சொல்லலாம்.

> [!NOTE]
> உங்கள் API கீவை உங்கள் கோடு முதல் பிரிக்க பரிந்துரைக்கப்படுகிறது. சுற்றுச் சூழல் மாறிகள் மூலம் இதை செய்யலாம்.
> - உங்கள் .env கோப்பில் சுற்றுச் சூழல் மாறி `AZURE_OPENAI_API_KEY`-யை உங்கள் API கீ ஆக அமைக்கவும். நீங்கள் இதுவரை இந்த பாடநெறியின் முன்னைய பயிற்சிகளை முடித்திருந்தால் ஏற்பாடு செய்யப்பட்டுள்ளது.


### Azure இயக்கு அமைக்கவும்

நீங்கள் Azure OpenAI-ஐப் பயன்படுத்தினால், இயக்கு அமைக்க வேண்டிய விதமாகும். Responses API Azure OpenAI **v1 இறுகுகை** மூலம் வழங்கப்படுகிறது, ஆகவே `OpenAI` கிளையண்டை `<your-endpoint>/openai/v1/` என்று குறிக்கிறோம்:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

மேலே கீழ்காணும் அமைப்பை வைத்துள்ளோம்:

- `api_key`, இது உங்கள் Azure போர்டலில் உள்ள API கீயே.
- `base_url`, இது உங்கள் Azure OpenAI இறுகுகை, இதில் `/openai/v1/` இணைக்கப்பட்டுள்ளது. நீங்கள் Azure போர்டலில் API கீ அருகிலுள்ள இறுகுகையை கண்டுபிடிக்கலாம். v1 இறுகுகை பயன்படுத்துவதால் `api_version` வழங்க தேவையில்லை.
- `deployment`, இது Foundry போர்டலில் நீங்கள் உருவாக்கிய மாதிரி இயக்கு பெயர்.

1. ஏற்கனவே உள்ள *app.py* கோப்பைப் பயன்படுத்தி தொடங்கவும்

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
