## 4. விருப்ப வடிவமைப்பு  

"பெரிய மொழி மாதிரிகளின் மாயாஜாலம் என்னவென்றால், அதிக அளவிலான உரை மூலம் இந்த முன்னறிவிப்பு பிழையை குறைப்பதற்காக பயிற்சி செய்யப்படுவதால், இந்த மாதிரிகள் இந்த முன்னறிவிப்புகளுக்கு பயன்படும் கருத்துக்களை கற்றுக்கொள்கின்றன. உதாரணமாக, அவை கீழ்க்கண்டவைகளைப் போன்ற கருத்துக்களை கற்றுக்கொள்ளுகின்றன"(1):

* எப்படி வரியில் எழுதுவது
* व्याकरणம் எப்படி வேலை செய்கிறது
* எப்படி வேறுபடுத்தி சொல்லுவது
* கேள்விகளுக்கு எப்படி பதில் கூறுவது
* உரையாடலை எப்படி நடத்துவது
* பல மொழிகளில் எப்படி எழுதுவது
* எப்படி குறியிடுவது
* போன்றவை.

#### பெரிய மொழி மாதிரியை எப்படி கட்டுப்படுத்துவது  
"பெரிய மொழி மாதிரிக்கு உள்ள அனைத்து உள்ளீடுகளிலும், மிகவும் பாதிப்பு அளிக்கிறது உரை விருப்பம்(1).

பெரிய மொழி மாதிரிகள் சில வழிகளில் வெளியீடு உற்பத்தி செய்ய விருப்பப்படுத்தப்படலாம்:

வழிமுறைகள்: மாதிரிக்கு நீங்கள் என்ன விரும்புகின்றீர்கள் என்பது சொல்லுங்கள்
நிறைவு: நீங்கள் விரும்பும் ஆரம்பத்தை மாதிரியில் நிறைவு செய்ய செய்யும்
காட்சி: நீங்கள் என்ன விரும்புகின்றீர்கள் என்பதை மாதிரிக்கு காட்டுங்கள், அல்லது:
விருப்பத்தில் சில எடுத்துக்காட்டுகள்
சிறந்தத் தனிமைப்படுத்தல் பயிற்சி தரவுத்தொகுப்பில் நூற்றுக்கணக்கான அல்லது ஆயிரக்கணக்கான எடுத்துக்காட்டுகள்"



#### விருப்பங்களை உருவாக்குவதற்கான மூன்று அடிப்படைக் வழிகாட்டுதல்கள் உள்ளன:

**கூறவும் மற்றும் காட்டவும்**. நீங்கள் என்ன விரும்புகிறீர்கள் என்பதைக் கட்டளை, எடுத்துக்காட்டுகள், அல்லது இரண்டின் கலவையினை மூலம் தெளிவாகக் கூறுங்கள். மாதிரி பத்தியை உணர்வுபூர்வமாக வகைப்படுத்த அல்லது உருப்படியை வரிசைப்படி வரிசைப்படுத்த விரும்பினால், அதுவே நீங்கள் விரும்புவதாக அதற்கு காண்பியுங்கள்.

**நன்கு தரமான தரவைக் கொடுங்கள்**. நீங்கள் ஒரு வகைப்படுத்தக்கருவியை உருவாக்க முயற்சிக்கிறீர்களோ அல்லது மாதிரியை ஒரு மாதிரியை பின்பற்றச் செய்ய விரும்பினாலும், போதுமான எடுத்துக்காட்டுகள் தகுந்திருக்க வேண்டும். உங்கள் எடுத்துக்காட்டுகளை பரிசீலித்து விட்டு பிழைகள் இல்லாதவாறே உறுதி செய்யுங்கள் — மாதிரி சாதாரண சொல்வழு பிழைகளை கண்டிருக்கும் அளவுக்கு புத்திசாலித்தனமாக இருக்கும் மற்றும் பதில் அளிக்கும், ஆனால் அது சில சமயங்களில் இது குறித்துத் திட்டமிட்டது என்று எடுத்துக்கொள்ளலாம், இது பதிலுக்கு தாக்கத்தை ஏற்படுத்தலாம்.

**உங்கள் அமைப்புகளை சரிபார்க்கவும்.** வெப்பநிலை மற்றும் top_p அமைப்புகள் மாதிரி பதில் உருவாக்குவதில் எவ்வளவு நிர்ணயமானது என்பதை கட்டுப்படுத்துகின்றன. நீங்கள் ஒரு பதிலுக்கு ஒரே சரியான பதில் வேண்டுமானால், இந்த அமைப்புகளை குறைக்க விரும்புவீர்கள். பல்வேறு பதில்கள் தேவைப்பட்டால், இவற்றை உயர்த்த விரும்பலாம். இந்த அமைப்புகளைப் பயன்படுத்துவதில் பொதுவான ஒரு பிழை என்னவென்றால் அவை "அறிவுப்படிப்பிறப்பு" அல்லது "புதுமை" கட்டுப்பாடுகள் என்று கருதுவது.

### 5. சமர்ப்பிக்கவும்!
