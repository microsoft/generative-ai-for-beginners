## 4. பதிலளிப்பு வடிவமைப்பு  

"பெரிய மொழி மாதிரிகளின் மாயாஜாலம் என்னவெனில், பெரும் அளவிலான உரைகளில் இந்த முன்கூட்டிய தவறு குறைக்க பயிற்சி பெற்றதால், இந்த மாதிரிகள் இந்த முன்னறிவிப்புகளுக்கு பயனுள்ள கொள்கைகளை கற்றுக்கொள்ளும். உதாரணமாக, அவைகள் இதுபோன்ற கொள்கைகளை கற்றுக்கொள்ளும்"(1):

* எப்படி எழுத்துப்பிழைகளை சரிசெய்ய வேண்டும்
* வாகியியல் எப்படி வேலை செய்கிறது
* எப்படி மறுஅர்த்தப்படுத்த வேண்டும்
* எப்படி கேள்விகளுக்கு பதில் அளிக்க வேண்டும்
* எப்படி உரையாடல் நடத்த வேண்டும்
* பல மொழிகளில் எப்படி எழுத வேண்டும்
* எப்படி குறியீடு (code) செய்ய வேண்டும்
* முதலியன.

#### ஒரு பெரிய மொழி மாதிரியை எப்படி கட்டுப்படுத்துவது  
"ஒரு பெரிய மொழி மாதிரிக்கான அனைத்து உள்ளீடுகளிலும், மிகப்பெரிய தாக்கம் பயனீடு செய்யும் ஒரு உரை பதிலளிப்பு(1).

பெரிய மொழி மாதிரிகளை சில வழிகளில் வெளியீடு செய்ய நிரப்பலாம்:

வழிகாட்டி: மாதிரிக்கு நீங்கள் என்ன வேண்டும் என்று சொல்லுங்கள்
நிறைவு: நீங்கள் வேண்டும் என்று ஆரம்பத்தை மாதிரிக்கு நிறைவு செய்யவழியுங்கள்
காட்சித்தொடர்: நீங்கள் என்ன வேண்டும் என்பதை மாதிரிக்கு காண்பியுங்கள், அவற்றில் ஒன்று:
பதிலளிப்பில் சில உதாரணங்கள்
சிறப்பு-பயிற்சி தரவுத்தொகுப்பில் நூற்றுக்கணக்கான அல்லது ஆயிரக்கணக்கான உதாரணங்கள்"



#### பதிலளிப்புகளை உருவாக்க மூன்று அடிப்படையான வழிகாட்டுதல்கள் உள்ளன:

**காட்டு மற்றும் சொல்லுங்கள்**. நீங்கள் என்ன வேண்டும் என்பதை வழிகாட்டுதல்கள், உதாரணங்கள் அல்லது இரண்டின் கலவையால் தெளிவாகச் சொல்லுங்கள். நீங்கள் மாதிரியில் பொருள்களை வரிசைப்படுத்தவும் அல்லது ஒரு பத்தியை உணர்வு படி வகைப்படுத்தவோ விரும்பினால், அது நீங்கள் வேண்டும் என்காட்டுங்கள்.

**தரமான தரவுகளை வழங்குங்கள்**. நீங்கள் வகைப்பாணி உருவாக்கவோ அல்லது மாதிரியை ஒரு முறைபாட்டை பின்பற்றவோ முயற்சித்தால், போதுமான உதாரணங்கள் இருப்பதாக உறுதிப்படுத்துங்கள். உங்கள் உதாரணங்களை சரிபார்க்க மறவாதீர்கள் — மாதிரி சாதாரண எழுத்துப்பிழைகள் மூலம் பதிலை வழங்க சிறிது முடிவு முடியும், ஆனால் அது இந்த தவறு உளைந்த தவறாக இருக்கலாம் என்றும் எண்ணக்கூடும், இது பதிலுக்கு பாதிப்பு ஏற்படுத்தும்.

**உங்கள் அமைப்புகளை சரிபாருங்கள்**. "temperature" மற்றும் "top_p" அமைப்புகள் மாதிரி பதில் உருவாக்கும் முறையில் எவ்வளவு குறிக்கொருந்தியதாக இருக்கும் என்பதை கட்டுப்படுத்தும். நீங்கள் ஒரு சரியான பதிலை வேண்டும் என கேட்டால், இதை குறைத்து அமைக்கலாம். மேலதிக வித்தியாசமான பதில்கள் விரும்பினால், இதை அதிகரிக்கலாம். இந்த அமைப்புகளில் மக்கள் செய்யும் முக்கிய தவறு, அவை "அறிவுத்திறன்" அல்லது "புதிய சிந்தனை" கட்டுப்பாடுகள் என்று நினைப்பது ஆகும்.

### 5. சமர்ப்பிக்கவும்!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
