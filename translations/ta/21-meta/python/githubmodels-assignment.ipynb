இவை லாமா 3.1 ஐ GenAI பயன்பாடுகள் கட்டமைப்பதில் சிக்கலான வழக்குகளை கையாள உதவுகின்றன, அதில்: 
- உள்ளூர்ப் பணி அழைப்பு - LLM வேலைப்பாட்டை தவிர்ந்த வெளிப்பக்க கருவிகள் மற்றும் பணிகளுக்கு அழைக்க உத்தி 
- சிறந்த RAG செயல்திறன் - அதிக உள்ளடக்க சாளரத்தின் பேரில் 
- செயற்கை தரவு உருவாக்கம் - நுண்ணறிவுக்கு போன்ற பணிகளுக்கு பயனுள்ள தரவு உருவாக்கும் திறன் 



### பிறமொழி செயற்பாட்டின் அழைப்பு  

Llama 3.1 செயல்பாடுகள் அல்லது கருவிகள் அழைப்புகளில் அதிக பயனுள்ளதாக இருக்க புதுப்பிக்கப்பட்டுள்ளது. இது இரண்டு உள்ளமைவுத் திறன்கள் கொண்டுள்ளன, அவை பயனரின் குறிப்பின் அடிப்படையில் பயன்படுத்த வேண்டியதாக கருதப்படுகின்றன. இவை: 

- **Brave Search** - இணைய தேடலை செய்ய வளியான தகவல்களைப் பெற பயன்படும், உதாரணமாக வானிலை 
- **Wolfram Alpha** - சிக்கலான கணித கணக்கீடுகளுக்குப் பயன்படுத்திக்கொள்ளலாம், எனவே உங்களுடையதான செயல்பாடுகளை எழுத தேவையில்லை. 

நீங்கள் உங்கள் தனிப்பட்ட கருவிகளை உருவாக்கி LLM அழைக்க வைத்துக் கொள்ளலாம். 

கீழுள்ள குறியீட்டு உதாரணத்தில்: 

- கிடைக்கும் கருவிகள் (brave_search, wolfram_alpha) சிஸ்டம் குறிப்பில் வரையறுக்கப்பட்டுள்ளன. 
- ஒரு நகரின் வானிலை பற்றி கேட்டு பயனர் குறிப்பை அனுப்புகிறார். 
- LLM Brave Search கருவிக்கு தொழில்நுட்ப அழைப்பை செய்யும், இது இப்படியே இருக்கும் `<|python_tag|>brave_search.call(query="Stockholm weather")` 

*குறிப்பு: இந்த உதாரணம் கருவி அழைப்பையே செய்கிறது, நீங்கள் முடிவுகளைப் பெற விரும்பினால், Brave API பக்கத்தில் ஒரு இலவச கணக்கை உருவாக்கி செயல்பாட்டையும் வரையறுக்க வேண்டும்* 


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### Llama 3.2 

ஒரு LLM ஆக இருந்தாலும், Llama 3.1 உடைய ஒரு குறைபாடு என்னவென்ப하면, அது பல் வடிவத்தன்மை ஆகும். அதாவது, பிரதிகளாக படங்கள் போன்ற வேறுபட்ட வகை உள்ளீடுகளைப் பயன்படுத்தி பதில்களை வழங்கும் திறன். இந்தக் கூற்று Llama 3.2 இன் முக்கிய அம்சங்களில் ஒன்றாகும். இக்கூற்றுகளுக்கும் பின்வரும் அம்சங்கள் சேர்க்கப்பட்டுள்ளன: 

- பல் வடிவத்தன்மை - உரை மற்றும் படம் ஆகிய இரு பிரதிகளையும் மதிப்பாய்வதற்கான திறன் 
- சிறிய முதல் நடுத்தர அளவுகள் (11B மற்றும் 90B) - இது நெகிழ்வான செயல்படுத்தும் விருப்பங்களை வழங்குகிறது, 
- உரை மட்டுமே நிலைகள் (1B மற்றும் 3B) - இது மாடலை எட்ஜ் / மொபைல் சாதனங்களில் இயக்க உதவுகிறது மற்றும் குறைந்த தாமதத்தைக் கொண்டுள்ளது 

பல் வடிவத்தன்மை ஆதரவு திறந்த மூல மாடல்களின் உலகில் ஒரு பெரிய முன்னேற்றமாகும். கீழே உள்ள குறியீட்டு எடுத்துக்காட்டில் படமும் உரை பிரதியும் இரண்டும் எடுத்து Llama 3.2 90B இல் இருந்து அந்த படத்தின் பகுப்பாய்வை பெறுகின்றது. 

### Llama 3.2 உடன் பல் வடிவத்தன்மை ஆதரவு


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
