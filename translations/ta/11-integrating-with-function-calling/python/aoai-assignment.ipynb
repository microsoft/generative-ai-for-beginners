இந்த பாடத்தில் நீங்கள் ஏதேனும் வேறு பாடத்தை முடித்திருந்தால், பெரிய மொழி மாதிரிகளின் (LLMs) சக்தியை நீங்கள் நிச்சயமாக புரிந்திருக்கிறீர்கள். வாய்ப்பளவில், அவற்றின் சில எல்லைகளையும் நீங்கள் புரிந்து கொள்ளலாம். 

செயல்பாடு அழைப்புகள் Azure Open AI சேவையின் ஒரு அம்சமாக, பின்வரும் எல்லைகளை 극복 செய்யும்: 
1) ஒரே மாதிரியான பதில் வடிவம் 
2) ஒரு உரையாடல் சூழலில் பயன்பாட்டின் பிற செயல்முறைகள் தரவுகளைப் பயன்படுத்தும் திறன் 

செயல்பாடு அழைப்புக்கு முன்பு, LLM இன் பதில்கள் அமைப்பு இல்லாதவையும் ஒரே மாதிரியானதுமல்லாதவையும் இருந்தன. டெவலப்பர்கள் ஒவ்வொரு பதிலும் மாறுபாடுகளையும் கையாள வலுவான சரிபார்ப்பு குறியீட்டை எழுத வேண்டியிருந்தது. 

பயனர்கள் "ஸ்டாக்க்ஹோம் நகரில் தற்போதைய வானிலை என்ன?" போன்ற கேள்விகளுக்கு பதில் பெற முடியவில்லை. இதற்கு காரணம், மாதிரிகள் பயிற்சி பெற்ற தரவின் நேரத்திலேயே மட்டுப்படுத்தப்பட்டிருந்தது. 

கீழே உள்ள உதாரணம் இந்த பிரச்சினையை விளக்குகிறது:

```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


இந்த தரவை பார்ஸ் செய்ய LLM-க்கு இதனை அனுப்ப விரும்புகிறோம். பின்னர் இதை எங்கள் செயலியில் API-க்கு அனுப்ப அல்லது தரவுத்தளத்தில் சேமிக்க பயன்படுத்தலாம்.

நாம் எதை பற்றிய தகவலை அறிய விரும்புகிறோம் என்பதை LLM-க்கு அறிவிக்க ஒரே மாதிரியான இரண்டு ப்ராம்ட்களை உருவாக்குவோம்:


நாங்கள் இதனை எல்.எல்.எம்-க்கு அனுப்பி, எங்கள் தயாரிப்பிற்கு முக்கியமான பகுதிகளை பகுப்பாய்வு செய்ய விரும்புகிறோம். எனவே எல்.எல்.எம்-க்கு வழிமொழிகள் அளிக்க இரண்டு தனித்துவமான அழைப்புகளை உருவாக்க முடியும்: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

இந்த இரண்டு ப்ராம்டுகளையும் உருவாக்கிய பிறகு, நாங்கள் அவற்றை `client.responses.create` ஐ பயன்படுத்தி LLMக்கு அனுப்புவோம்.  ப்ராம்டை `input` மாறியில் சேமித்து, `user` என்ற வேடத்துக்கு ஒதுக்குகிறோம். இது ஒரு பயனரின் செய்தியை ஒரு சாதாரண உரையாடல் பொறிக்கு எழுதப்படுவதைப் போல கற்பனை செய்யும். 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

இப்போது நாம் இரு கோரிக்கைகளையும் LLM-க்கு அனுப்பிவிட்டு, பெறும் பதிலை பரிசோதிக்கலாம். 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```

## 2. உங்கள் முதல் செயல்பாடுத் தொடர்பை உருவாக்குதல் 

ஒரு செயல்பாடுத் தொடர்பை உருவாக்கும் செயல்முறை 3 முக்கிய படிகள் கொண்டது: 
1. உங்கள் செயல்பாடுகள் மற்றும் பயனர் செய்தி கொண்ட பட்டியலுடன் Chat Completions API ஐ அழைப்பது 
2. ஒரு செயல்முறை அல்லது API அழைப்பை நடத்துவதற்கான நடவடிக்கை மேற்கொள்ள மாடல் பதிலைப் படித்தல் 
3. உங்கள் செயல்பாடின் பதிலுடன் Chat Completions APIயை மீண்டும் அழைத்து, அந்தத் தகவலைப் பயன்படுத்தி பயனருக்கான பதிலை உருவாக்குதல்.

### ஒரு செயல்பாட்டு அழைப்பின் கூறுகள் 

#### பயனர்கள் உள்ளீடு 

முதல் படி ஒரு பயனர் செய்தியை உருவாக்குதல். இது ஒரு உரை உள்ளீட்டின் மதிப்பை எடுத்து இயக்கமாக வழங்க முடியும் அல்லது நீங்கள் இங்கே ஒரு மதிப்பை வழங்கலாம். இது Chat Completions API-யுடன் முதன்முறையாக வேலை செய்கிறீர்களானால், நாம் `role` மற்றும் `content` என்பதை வரையறுக்க வேண்டும். 

`role` என்பது `system` (விதிகளை உருவாக்குதல்), `assistant` (மாதிரி) அல்லது `user` (இறுதி பயனர்) ஆக இருக்கலாம். செயல்பாட்டு அழைப்புக்கு, நாம் இதை `user` மற்றும் ஒரு உதாரண கேள்வியாக அமைப்போம். 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### செயல்பாடுகளை உருவாக்குதல். 

அடுத்து, நாமு ஒரே செயல்பாடு மற்றும் அந்த செயல்பாட்டின் பாறமிட்டர்களைப் 정의 செய்ய போகிறோம். இங்கே `search_courses` என்ற செயல்பாடையே நாம்பு பயன்படுத்தப் போகிறோம், ஆனால் நீங்கள் பல செயல்பாடுகளை உருவாக்கலாம்.

**முக்கியம்** : செயல்பாடுகள் LLMக்கு அனுப்பப்படும் அமைப்பு செய்தியில் சேர்க்கப்படுகின்றன மற்றும் நீங்கள் பயன்படுத்தக்கூடிய வாழ்ந்துள்ள டோகன்களின் தொகையிலும் சேரும். 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**வரையறைகள்** 

`name` - நாம்பதற்கு அழைக்க வேண்டும் எனக் கோரப்படும் செயலாக்கத்தின் பெயர். 

`description` - செயலாக்கம் எப்படிச் செயல்படுகிறது என்பதற்கான விளக்கம். இங்கே தெளிவாகவும் குறிப்பாகவும் இருக்க வேண்டும் 

`parameters` - மாடல் அதன் பதிலில் உருவாக்க வேண்டிய மதிப்புகளையும் வடிவத்தையும் உள்ளடக்கிய பட்டியல் 


`type` - பண்புகளின் தரவுப் பிரிவு சேமிக்கப்படும் 

`properties` - மாடல் பதிலுக்காக பயன்படுத்தும் குறிப்பிட்ட மதிப்புகளின் பட்டியல் 


`name` - மாடல் வடிவமைக்கப்பட்ட பதிலில் பயன்படுத்தும் பண்பின் பெயர் 

`type` - இந்நிறுவனத்தின் தரவுப் பிரிவு 

`description` - குறிப்பிட்ட பண்பின் விளக்கம் 


**விருப்பமானது**

`required` - செயலாக்கக் கோரிக்கை நிறைவேற வேண்டிய அவசியமான பண்பு 


### செயல்பாடை அழிப்பது  
ஒரு செயல்பாட்டை வரையறுத்த பிறகு, அதை Chat Completion API கCallingல் சேர்க்க வேண்டும். இதை `functions` ஐ கோரிக்கையில் சேர்ப்பதன் மூலம் செய்கிறோம். இந்த வழக்கில் `functions=functions`.  

`function_call` ஐ `auto` ஆக அமைப்பதற்கும் ஒரு விருப்பம் உள்ளது. இது பயனர் செய்திக்கு அடிப்படையாய் எந்த செயல்பாடு அழைக்கப்பட வேண்டும் என்பதை LLM தான் தீர்மானிக்க விடுவோம் என பொருள்.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

இப்போது பதிலைக் காண்போம் மற்றும் அது எவ்வாறு வடிவமைக்கப்பட்டுள்ளது என்பதைப் பார்ப்போம்: 

{
  "role": "assistant",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

உங்கள் செய்தியிலிருந்து, LLM இந்த செயல்பாட்டு காரணிகளுக்கு பொருத்தமான தரவை கண்டுபிடிக்க முடிந்தது, செயல்பாட்டின் பெயர் அழைக்கப்படுவதை நீங்கள் காணலாம். 


## 3. செயலியில் செயல்பாடு அழைப்புகளை இணைக்கும் போது. 


LLM இலிருந்து வடிவமைக்கப்பட்ட பதிலை நாம் சோதித்த பிறகு, தற்போது இதை ஒரு செயலியில் இணைக்கலாம். 

### ஓட்டத்தை நிர்வகித்தல் 

இதை நமது செயலியில் இணைக்க, கீழ்காணும் படிகளை பின்பற்றுவோம்: 

முதலில், Open AI சேவைகளுக்கு அழைப்பு செய்யவும், மற்றும் செய்தியை `response_message` என்ற மாறியில் சேமிக்கவும். 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

இப்போது நாம் மைக்ரோசாஃப்ட் லேர்ன் API-ஐ அழைத்து படிப்புகளின் பட்டியலை பெறும் செயற்பாட்டை வரையறுப்போம்: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

சிறந்த நடைமுறையாக, மாடல் ஒரு செயல்பாட்டை அழைக்க விரும்புகிறதா என்பதை நாம் பார்ப்போம். அதன் பிறகு, வருவன செயல்பாடுகளில் ஒன்றை உருவாக்கி, அழைக்கபடும் செயல்பாட்டுடன் பொருந்தும் வகையில் அதனை பொருத்துவோம்.
பின்னர் செயல்பாட்டின் аргументுகளை எடுத்து, அவற்றை LLM இல் இருந்து வரும் аргументுகளுடன் பொருத்துவோம்.

கடைசியாக, `search_courses` செய்தியால் திரும்பப்பட்ட மதிப்புகளுடன் செயல்பாட்டு அழைப்புச் செய்தியை இணைக்கிறோம். இதன் மூலம் LLMக்கு அனைத்து தேவையான தகவல்களும் கிடைக்கின்றன
இயல்பான மொழியில் பயனருக்கு பதிலளிக்க.


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

இப்போது நாம் புதுப்பிக்கப்பட்ட செய்தியை LLMக்கு அனுப்பப் போகிறோம் என்பதால், API JSON வடிவமைக்கப்பட்ட பதிலுக்கு பதிலாக ஒரு இயல்பான மொழி பதிலை பெற முடியும்.
