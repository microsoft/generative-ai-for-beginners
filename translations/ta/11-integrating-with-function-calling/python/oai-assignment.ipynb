செயல்பாடு அழைப்பு என்பது OpenAI சேவையின் ஒரு அம்சமாகும், இது பின்வரும் சவால்களை சமாளிக்க வடிவமைக்கப்பட்டுள்ளது:

ஒற்றுமையற்ற பதில் வடிவமைப்பு:
- செயல்பாடு அழைப்புக்கு முன்பு, பெரிய மொழி மாதிரியில் இருந்து வரும் பதில்கள் அமைப்பற்றதும் ஒற்றுமையற்றதும் இருந்தன. ஒவ்வொரு வெளியீட்டு மாறுபாட்டையும் கையாளுவதற்கு டெவலப்பர்கள் சிக்கலான சரிபார்ப்பு குறியீட்டை எழுத வேண்டியிருந்தது.

வெளிப்புற தரவுடன் குறைந்த ஒருங்கிணைப்பு:
- இந்த அம்சத்திற்கு முன்பு, ஒரு செயலியில் உள்ள மற்ற பகுதிகளிலிருந்து தரவை உரையாடல் சூழலில் சேர்ப்பது கடினமாக இருந்தது.

பதில்களின் வடிவங்களை ஒரே மாதிரியாக்கி, வெளிப்புற தரவுடன் எளிதாக ஒருங்கிணைக்க செயல்பாடு அழைப்பு வளர்ச்சியை எளிதாக்கி கூடுதல் சரிபார்ப்பு தர்க்கத்தின் தேவையை குறைக்கிறது.

பயனர்கள் "ஸ்டாக்ஹோல்ம் இல் தற்போதைய வானிலை என்ன?" போன்ற பதில்களை பெற முடியவில்லை. இதன் காரணம், மாதிரிகள் பயிற்சி பெற்ற தரவின் காலத்திற்கே வரம்பு கொண்டிருந்தது.

நாம் மாணவர்களின் தரவுத்தளத்தை உருவாக்கி, அவர்களுக்கு சரியான பாடத்திட்டத்தை பரிந்துரைக்க விரும்புவோம் என்று கூறுவோம். கீழே உள்ள இரண்டு மாணவர்களின் விவரங்கள் தரவில் மிகவும் ஒத்ததாக உள்ளன.

நாம் இந்தத் தரவுகளை பகுப்பாய்வு செய்ய LLM-க்கு அனுப்ப விரும்புகிறோம். இதை பின்னர் எங்கள் பயன்பாட்டில் API-க்கு அனுப்ப அல்லது தரவுத்தளத்தில் சேமிக்க பயன்படுத்தலாம்.

நாம் ஆர்வமாக உள்ள தகவல்களை LLM-க்கு அறிவிக்க இரண்டு ஒரே மாதிரியான கேள்விகளை உருவாக்குவோம்:


நாம் இதை எங்கள் தயாரிப்புக்கு முக்கியமான பகுதிகளை பகுப்பாய்வு செய்ய LLMக்கு அனுப்ப விரும்புகிறோம். ஆகவே, LLMக்கு வழிகாட்ட இரண்டு ஒரே மாதிரியான கேள்விப்பட்டிகளை உருவாக்கலாம்:


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

இந்த இரண்டு ப்ராம்ட்களையும் உருவாக்கிய பிறகு, அவற்றை `openai.ChatCompletion` பயன்படுத்தி LLMக்கு அனுப்புவோம். ப்ராம்ட் `messages` என்ற மாறியில் சேமிக்கப்படுகிறது மற்றும் பங்கு `user` ஆக ஒதுக்கப்படுகிறது. இது ஒரு பயனர் ஒரு சாட்பாட்டுக்கு எழுதும் செய்தியை போல உருவாக்குவதற்காக ஆகும்.


```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-3.5-turbo"
```

இப்போது நாம் இரு கோரிக்கைகளையும் LLMக்கு அனுப்பி, பெறும் பதிலை பரிசீலிக்கலாம்.


```python
openai_response1 = client.chat.completions.create(
 model=deployment,    
 messages = [{'role': 'user', 'content': prompt1}]
)
openai_response1.choices[0].message.content 
```

```python
openai_response2 = client.chat.completions.create(
 model=deployment,    
 messages = [{'role': 'user', 'content': prompt2}]
)
openai_response2.choices[0].message.content
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.choices[0].message.content)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.choices[0].message.content )
json_response2
```

## 2. உங்கள் முதல் செயல்பாட்டு அழைப்பை உருவாக்குதல்

ஒரு செயல்பாட்டு அழைப்பை உருவாக்கும் செயல்முறை 3 முக்கிய படிகளைக் கொண்டுள்ளது:  
1. உங்கள் செயல்பாடுகளின் பட்டியலுடன் மற்றும் ஒரு பயனர் செய்தியுடன் Chat Completions API ஐ அழைக்கவும்  
2. ஒரு செயல்பாட்டை அல்லது API அழைப்பை செயல்படுத்த மாடலின் பதிலை படிக்கவும்  
3. பயனருக்கு பதிலை உருவாக்க அந்த தகவலை பயன்படுத்த உங்கள் செயல்பாட்டின் பதிலுடன் Chat Completions API ஐ மீண்டும் அழைக்கவும்.

### ஒரு செயல்பாட்டு அழைப்பின் கூறுகள்

#### பயனரின் உள்ளீடு

முதல் படி ஒரு பயனர் செய்தியை உருவாக்குவதாகும். இது ஒரு உரை உள்ளீட்டின் மதிப்பை எடுத்துக்கொண்டு தானாக ஒதுக்கப்படலாம் அல்லது நீங்கள் இங்கே ஒரு மதிப்பை ஒதுக்கலாம். இது Chat Completions API உடன் உங்கள் முதல் முறையான பணியாயின் போது, செய்தியின் `role` மற்றும் `content` ஐ வரையறுக்க வேண்டும்.

`role` என்பது `system` (விதிகளை உருவாக்குதல்), `assistant` (மாதிரி) அல்லது `user` (இறுதி பயனர்) ஆக இருக்கலாம். செயல்பாட்டு அழைப்புக்கு, இதை `user` ஆகவும் ஒரு உதாரணக் கேள்வியாகவும் ஒதுக்குவோம்.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### செயல்பாடுகளை உருவாக்குதல்.

அடுத்து நாம் ஒரு செயல்பாட்டையும் அந்த செயல்பாட்டின் அளவுருக்களையும் வரையறுப்போம். இங்கே `search_courses` என்ற ஒரு செயல்பாட்டை மட்டுமே பயன்படுத்துவோம், ஆனால் நீங்கள் பல செயல்பாடுகளை உருவாக்கலாம்.

**முக்கியம்** : செயல்பாடுகள் LLM க்கு அனுப்பப்படும் அமைப்பு செய்தியில் சேர்க்கப்படுகின்றன மற்றும் நீங்கள் பயன்படுத்தக்கூடிய டோக்கன்களின் அளவிலும் சேர்க்கப்படும்.


```python
functions = [
   {
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**வரையறைகள்** 

செயல்பாட்டு வரையறை அமைப்பில் பல நிலைகள் உள்ளன, ஒவ்வொன்றும் தன் சொந்த பண்புகளைக் கொண்டுள்ளது. இங்கே அந்த நெஸ்டட் அமைப்பின் பகிர்வு:

**மேல் நிலை செயல்பாட்டு பண்புகள்:**

`name` - நாம் அழைக்க விரும்பும் செயல்பாட்டின் பெயர்.

`description` - செயல்பாடு எப்படி செயல்படுகிறது என்பதற்கான விளக்கம். இங்கே தெளிவாகவும் குறிப்பாகவும் இருக்க வேண்டும்.

`parameters` - மாடல் தனது பதிலில் உருவாக்க விரும்பும் மதிப்புகளும் வடிவத்தையும் கொண்ட பட்டியல்.

**பராமரிப்பு பொருள் பண்புகள்:**

`type` - பராமரிப்பு பொருளின் தரவுத் தளம் (பொதுவாக "object")

`properties` - மாடல் தனது பதிலுக்காக பயன்படுத்தும் குறிப்பிட்ட மதிப்புகளின் பட்டியல்.

**தனிப்பட்ட பராமரிப்பு பண்புகள்:**

`name` - பண்புக் குறியீட்டால் மறைமுகமாக வரையறுக்கப்படுகிறது (எ.கா., "role", "product", "level")

`type` - இந்த குறிப்பிட்ட பராமரிப்பின் தரவுத் தளம் (எ.கா., "string", "number", "boolean")

`description` - குறிப்பிட்ட பராமரிப்பின் விளக்கம்.

**விருப்ப பண்புகள்:**

`required` - செயல்பாட்டை நிறைவேற்ற தேவையான பராமரிப்புகளை பட்டியலிடும் வரிசை.


### செயல்பாட்டு அழைப்பை உருவாக்குதல்  
ஒரு செயல்பாட்டை வரையறுத்த பிறகு, அதை Chat Completion API க்கு அழைப்பில் சேர்க்க வேண்டும். இதை செய்ய, கோரிக்கையில் `functions` ஐ சேர்க்கிறோம். இந்த நிலையில் `functions=functions` ஆகும்.  

`function_call` ஐ `auto` ஆக அமைக்கும் விருப்பமும் உள்ளது. இதன் பொருள், பயனர் செய்தியின் அடிப்படையில் எந்த செயல்பாடு அழைக்கப்பட வேண்டும் என்பதை நாங்கள் தீர்மானிப்பதற்குப் பதிலாக LLM தீர்மானிக்க விடுவோம் என்பதாகும்.


```python
response = client.chat.completions.create(model=deployment, 
                                        messages=messages,
                                        functions=functions, 
                                        function_call="auto") 

print(response.choices[0].message)
```

## 3. செயலியில் செயல்பாட்டு அழைப்புகளை ஒருங்கிணைத்தல். 


LLM இல் இருந்து வடிவமைக்கப்பட்ட பதிலை நாம் சோதித்த பிறகு, இப்போது இதை ஒரு செயலியில் ஒருங்கிணைக்கலாம். 

### ஓட்டத்தை நிர்வகித்தல் 

இதனை நமது செயலியில் ஒருங்கிணைக்க, கீழ்க்கண்ட படிகளை எடுத்துக்கொள்வோம்: 

முதலில், OpenAI சேவைகளுக்கு அழைப்பு செய்து, செய்தியை `response_message` என்ற மாறியில் சேமிப்போம். 


```python
response_message = response.choices[0].message
```

இப்போது நாம் Microsoft Learn API ஐ அழைத்து பாடநெறிகளின் பட்டியலை பெறும் செயல்பாட்டை வரையறுக்கப்போகிறோம்:


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ஒரு சிறந்த நடைமுறையாக, பின்னர் மாடல் ஒரு செயல்பாட்டை அழைக்க விரும்புகிறதா என்று பார்க்கப்போகிறோம். அதன்பின், கிடைக்கும் செயல்பாடுகளில் ஒன்றை உருவாக்கி, அழைக்கப்படும் செயல்பாட்டுடன் பொருந்தச் செய்வோம்.  
பின்னர், செயல்பாட்டின் வாதங்களை எடுத்துக் கொண்டு அவற்றை LLM இன் வாதங்களுடன் பொருத்துவோம்.

இறுதியில், செயல்பாடு அழைக்கும் செய்தியையும் `search_courses` செய்தியால் திரும்பிய மதிப்புகளையும் இணைப்போம். இது LLM க்கு இயல்பான மொழியைப் பயன்படுத்தி பயனருக்கு பதிலளிக்க தேவையான அனைத்து தகவல்களையும் வழங்குகிறது.


```python
# Check if the model wants to call a function
if response_message.function_call.name:
    print("Recommended Function call:")
    print(response_message.function_call.name)
    print()

    # Call the function. 
    function_name = response_message.function_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(response_message.function_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the assistant response and function response to the messages
    messages.append( # adding assistant response to messages
        {
            "role": response_message.role,
            "function_call": {
                "name": function_name,
                "arguments": response_message.function_call.arguments,
            },
            "content": None
        }
    )
    messages.append( # adding function response to messages
        {
            "role": "function",
            "name": function_name,
            "content":function_response,
        }
    )

```

இப்போது நாம் புதுப்பிக்கப்பட்ட செய்தியை LLM க்கு அனுப்புவோம், இதனால் நாம் API JSON வடிவமைக்கப்பட்ட பதிலுக்கு பதிலாக இயற்கை மொழி பதிலை பெற முடியும்.
