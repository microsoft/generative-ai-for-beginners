இதை முடிக்க நாம் சேர்த்து பயன்படுத்துவோம்:
 - பயனருக்கு ஒரு உரையாடல் அனுபவத்தை உருவாக்க `OpenAI`
 - பயனர்களுக்கு அவர்களது கோரிக்கையின் அடிப்படையில் பாடங்களைத் தேட உதவும் `Microsoft Learn Catalog API`
 - பயனரின் வினாவை எடுத்துக் கொண்டு API கோரிக்கையைக் செய்ய ஒரு செயல்பாட்டுக்கு அனுப்ப `Function Calling`

தொடங்க முதலில், நாங்கள் ஏன் செயல்பாடு அழைப்பை பயன்படுத்த விரும்புகிறோம் என்பதை பார்க்கலாம்:


### ஏன் Function Calling  

இந்தப் பாடத்திலிருந்து நீங்கள் ஏதேனும் பிற பாடங்களை முடித்திருந்தால், பெரிய மொழி மாதிரிகள் (LLMs) பயன்படுத்துவதின் சக்தியை நீங்கள் உணர்ந்து கொண்டிருக்க வாய்ப்பு உள்ளது. அவர்களின் சில வரம்புகளையும் நீங்கள் காண முடியும் என நம்புகிறோம். 

செயல்பாட்டு அழைப்பு என்பது OpenAI சேவையின் ஒரு அம்சமாக, கீழ்க்கண்ட சவால்களை எதிர்கொள்ள வடிவமைக்கப்பட்டுள்ளது:

மரபின்மை பதில் வடிவமைப்பு:
- செயல்பாட்டு அழைப்புக்கு முன்பு, பெரிய மொழி மாதிரியின் பதில்கள் ஒழுங்கற்ற மற்றும் ஒரே மாதிரியல்லாதவையாக இருந்தன. உருவாக்குனர்கள் ஒவ்வொரு மாற்றத்திற்கும் சற்று சிக்கலான சரிபார்ப்பு குறியீட்டைக் கட்டுவதாய் இருந்தனர்.

வெளிப்புற தரவு ஒருங்கிணைப்பில் வரம்புகள்:
- இந்த அம்சத்திற்கு முன்பு, செயலியில் இருந்த பிற பகுதிகளிலிருந்து தரவை உரையாடல் சூழலில் சேர்க்க கடினமாக இருந்தது.

பதில் வடிவங்களை ஒருங்கிணைத்தல் மற்றும் வெளிப்புற தரவுடன் துண்டா இணைப்பை வழங்குவதன் மூலம், செயல்பாட்டு அழைப்பு வளர்ச்சியை எளிதாக்கி கூடுதல் சரிபார்ப்பு லாஜிக் தேவையை குறைக்கிறது.

பயனாளிகள் "ஸ்டாக்ஹோல்ம் நகரில் தற்போதைய வானிலை என்ன?" போன்ற பதில்களை பெற முடியாது. இதன் காரணம், மாதிரிகள் பயிற்சி தரவின் காலம் வரை மட்டுப்பட்டவைகள்.

கீழ்கண்ட எடுத்துக்காட்டைப் பார்த்து இந்தச் சிக்கலை விளக்குவோம்:

நாம் மாணவர்களின் தரவுத்தளத்தை உருவாக்கி அவர்களுக்கு சரியான பாடநெறியை பரிந்துரைக்க விரும்புவோம் எனக் கற்பனை செய்க. கீழே இரு மாணவர்களின் விவரங்கள் உள்ளன, அவற்றில் உள்ள தரவு மிகவும் ஒத்தன.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


இந்த தரவை பகுப்பாய்வு செய்ய LLM-க்கு அனுப்ப விரும்புகிறோம். இதை பின்னர் எங்கள் செயலியில் API-க்கு அனுப்ப அல்லது தரவுத்தளத்தில் சேமிக்க பயன்படுத்தலாம்.

நாம் ஆர்வம் கொண்ட தகவல்களை LLM-க்கு அறிவிக்க இரண்டு ஒரேபோன்ற வழிமுறைகளை உருவாக்குவோம்:


எங்கள் தயாரிப்புக்கு முக்கியமான பகுதியை பகுப்பாய்வு செய்ய LLM-க்கு இதை அனுப்ப விரும்புகிறோம். எனவே LLM-ஐ வழிநடத்த இரண்டு ஒரே பதிப்பிலான அச்சுறுத்தல்களை உருவாக்க முடியும்: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

இந்த இரண்டு கோரிக்கைகளை உருவாக்கிய பிறகு, அவைகளை `client.responses.create` பயன்படுத்தி LLM க்கு அனுப்புவோம்.  நாம் கோரிக்கை `input` மாறியில் சேமித்து, பங்கு `user` ஆக ஒதுக்குகிறோம். இது ஒரு பயனர் மூலம் ஒரு செய்தி ஒரு chatbot க்கு எழுதப்படும் முறையை நகலெடுக்க உள்ளது.



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"
```

இப்போது நாம் இரண்டு கோரிக்கைகளையும் LLMக்கு அனுப்பி, பெறும் பதிலை பரிசீலிக்கலாம். 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```

## 2. உங்கள் முதல் செயல்பாட்டு அழைப்பை உருவாக்குதல் 

ஒரு செயல்பாட்டு அழைப்பு உருவாக்கும் செயல்முறை 3 முக்கிய அடுக்கு உள்ளன: 
1. உங்கள் செயல்பாடுகளின் பட்டியலும் பயனர் செய்தியும்கொண்ட Chat Completions APIஐ அழைப்பது 
2. செயல்பாடு அல்லது API அழைப்பை இயக்க ஒரு செயலைச் செய்ய மாடலின் பதிலைப் படிப்பது 
3. பயனருக்கு பதிலை உருவாக்க அந்த தகவலைப் பயன்படுத்தி உங்கள் செயல்பாட்டின் பதிலுடன் Chat Completions APIக்கு மற்றொரு அழைப்பை செய்வது.

### ஒரு செயல்பாட்டுக் கால் கூறுகள் 

#### பயனர் உள்ளீடு 

முதல் படி ஒரு பயனர் செய்தியை உருவாக்குவது. இது ஒரு எழுத்து உள்ளீட்டின் மதிப்பை எடுத்து தானாக ஒதுக்கப்படலாம் அல்லது நீங்கள் இங்கே ஒரு மதிப்பை ஒதுக்கலாம். இது Chat Completions API-உடன் உங்கள் முதல் முறையான பணியாயின் போது, `role` மற்றும் `content` என்பதை வரையறுக்க வேண்டும். 

`role` என்பது `system` (விதிகளை உருவாக்குதல்), `assistant` (மாதிரி) அல்லது `user` (இறுதி பயனர்) ஆகியவற்றில் ஒன்றாக இருக்கலாம். செயல்பாட்டுக் கால் செய்வதற்காக, இதை `user` என்றும் ஒரு உதாரணக் கேள்வியையும் ஒதுக்குவோம். 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### செயல்பாடுகளை உருவாக்குதல். 

அடுத்ததாகநாம் ஒரு செயல்பாடையும் அந்த செயல்பாடின் பராமரிப்புகளையும் வரையறுக்கப் போகிறோம். இங்கு `search_courses` என்ற ஒரு செயல்பாட்டை மட்டும் பயன்படுத்துகிறோம், ஆனால் நீங்களே பல செயல்பாடுகளை உருவாக்கலாம்.

**முக்கியம்** : செயல்பாடுகள் LLM க்கு அனுப்பப்படும் முறைச் செய்தியில் சேர்க்கப்பட்டுள்ளன மற்றும் நீங்கள் கொண்டுள்ள கிடைக்கும் டோக்கன்களின் அளவில் சேர்க்கப்படும். 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**வரையறைகள்** 

செயல்பாட்டு வரையறை கட்டமைப்பில் பல நிலைகள் உள்ளன, ஒவ்வொன்றிற்கும் தன் சொந்த பண்புகள் உள்ளன. இங்கே இந்த உள்நிலை கட்டமைப்பின் பகுப்பாய்வாக உள்ளது:

**மேல் நிலை செயல்பாட்டு பண்புகள்:**

`name` -  அழைக்க விரும்பும் செயல்பாட்டின் பெயர். 

`description` - செயல் எப்படி வேலை செய்கிறது என்பது குறித்த விளக்கம். இங்கு தெளிவாக மற்றும் குறிப்பிட்டவாறு இருக்க வேண்டும் 

`parameters` - மாதிரி தனது பதிலில் உருவாக்க விரும்பும் மதிப்புகளும் வடிவத்தையும் உள்ளடக்கிய பட்டியல் 

**பொருட்களின் பண்புகள்:**

`type` -  பொருள் வகையின் தரவு (சாதாரணமாக "object")

`properties` - மாதிரி தனது பதிலுக்கு பயன்படுத்தும் குறிப்பிட்ட மதிப்புகளின் பட்டியல் 

**தனித்த தனிப்பட்ட பண்புகள்:**

`name` - பண்பு திறவுகோலால் மறைமுகமாக நிர்ணயிக்கப்பட்டது (உதா., "role", "product", "level")

`type` - இந்த குறிப்பிட்ட பண்பின் தரவு வகை (உதா., "string", "number", "boolean") 

`description` - குறிப்பிட்ட பண்பின் விளக்கம் 

**விருப்பமான பண்புகள்:**

`required` - செயல்பாடு அழைப்பை முடிக்க எந்த பண்புகள் தேவையானவையென்று பட்டியலிடும் வரிசை 


### செயல்பாட்டை அழைக்கும் முறையை உருவாக்குதல்
ஒரு செயல்பாட்டை வரையறுத்த பிறகு, அதை இப்போது Chat Completion API-க்கு அழைப்பில் சேர்க்க வேண்டும். இதை, கோரிக்கையில் `functions` சேர்ப்பதன் மூலம் செய்வோம். இந்தக் கட்சியில் `functions=functions`.

`function_call` ஐ `auto` என அமைக்கவும் ஒரு விருப்பம் உள்ளது. இதன் பொருள் என்னவெனில், நாங்களைத் தானாக நியமிக்காமல், பயனர் செய்தி அடிப்படையில் எந்த செயல்பாடு அழைக்கப்பட வேண்டும் என்று LLM தன்னிச்சையாக தீர்மானிக்கும்.


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

{

  "function_call": {
    "name": "search_courses",

  }
}

நீங்கள் காண்கிறீர்கள் என்றால், செயல்பாட்டின் பெயர் அழைக்கப்பட்டது மற்றும் பயனர் செய்தியிலிருந்து, LLM செயல்பாட்டின் இடைமுக அளவுருக்களை பூர்த்தி செய்யும் தரவை கண்டுபிடிக்க முடிந்தது. 


## 3. செயலியில் செயல்பாட்டு அழைப்புக்களை ஒருங்கிணைத்தல்.


LLM இருந்து வடிவமைக்கப்பட்ட பதிலை நாம் சோதித்துவிட்ட பிறகு, இப்போது நாம் இதனை ஒரு செயலியில் ஒருங்கிணைக்கலாம்.

### ஓட்டத்தை நிர்வகிப்பு

இதனை நமது செயலியில் ஒருங்கிணைக்க, கீழ்காணும் படிகளை எடுத்துக்கொள்வோம்:

முதலில், OpenAI சேவைகளுக்கு அழைக்கவும், அதன் செய்தியினை `response_message` என்ற மாறியில் சேமிப்போம்.


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

இப்போது Microsoft Learn API ஐ அழைத்து பாடநெறிகள் பட்டியலை பெறும் செயல்பாட்டை வரையறுக்கப்போகிறோம்: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

சிறந்த நடைமுறையாக, அந்த மாடல் ஒரு செயல்பாட்டை அழைக்க விரும்புகிறதா என்பது நாம் கடைசியில் பார்க்கப்போகிறோம். அதன்பிறகு, நாம_available_ செயல்பாடுகளின் ஒன்றை உருவாக்கி, அது அழைக்கப்படும் செயல்பாட்டுடன் பொருந்தும் என்று பொருந்துகிறோம்.
பின்னர், செயல்பாட்டின் аргументы களை எடுத்து, அவை LLM அமர்த்தப்பட்ட аргументы களுடன் பொருந்தும்.

இறுதியில், செயல்பாட்டு அழைப்பு செய்தியை மற்றும் `search_courses` செய்தியால் திருப்பிய மதிப்புகளை இணைக்கப்போகிறோம். இது LLM க்கு தேவையான அனைத்து தகவல்களையும் வழங்கி
இயற்கை மொழி பயன்படுத்தி பயனருக்கு பதிலளிக்க உதவுகிறது.
