### Build your first prompt  
Dis small exercise go give you basic intro on how to submit prompts to OpenAI model for simple waka "summarization".


**Steps**:  
1. Install OpenAI library for your python environment  
2. Load standard helper libraries and set your normal OpenAI security credentials for the OpenAI Service wey you don create  
3. Choose model wey you wan use for your task  
4. Create simple prompt for the model  
5. Submit your request to the model API!


### 1. Install OpenAI

### 2. Import helper libraries an set up credentials


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Finding di correct model  
Models like GPT-4o and GPT-4o mini sabi understand and generate natural language, and dem dey available for OpenAI platform with different levels of power and speed wey fit different kain work.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Prompt Design  

"Di magic wey dey big big language models be say as dem dey train am to reduce dis prediction error for plenti text, di models go learn concepts wey go useful for these predictions. For example, dem go learn concepts like"(1):

* how to spell
* how grammar dey work
* how to paraphrase
* how to answer questions
* how to hold conversation
* how to write for plenty languages
* how to code
* etc.

#### How to control big big language model  
"Among all di inputs wey go a big big language model, di text prompt na di one wey get di most influence(1).

Big big language models fit get prompted to generate output in some ways:

Instruction: Talk di model wetin you want
Completion: Make di model finish di beginning of wetin you want
Demonstration: Show di model wetin you want, with either:
Some examples for di prompt
Plenti hundreds or thousands examples inside fine-tuning training dataset"



#### Dem get three basic guidelines to create prompts:

**Show and tell**. Make am clear wetin you want whether na instructions, examples, or both. If you want di model make e arrange list of items alphabetically or classify paragraph by sentiment, show am say na wetin you want.

**Provide quality data**. If you dey try build classifier or want make di model follow one pattern, make sure say examples dey enough. Make you proofread your examples — di model sabi to see basic spelling mistakes and still give you response, but e fit also think say na on purpose and e fit affect di response.

**Check your settings.** Di temperature and top_p settings control how sure di model be to generate answer. If you dey ask for answer wey get only one correct one, then you go want make dem low. If you dey find more diverse answers, then you fit set dem high. Di number one mistake wey people dey make for these settings na to think say na "cleverness" or "creativity" controls.


Source: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Submit!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

### If you call am again, how results dem go compare?


```python

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
