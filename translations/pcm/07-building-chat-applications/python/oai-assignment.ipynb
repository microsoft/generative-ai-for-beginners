### Build your first prompt  
Dis short exercise go give una basic intro for how to submit prompts to OpenAI model for simple work "summarization".  


**Steps**:  
1. Install OpenAI library for your python environment  
2. Load standard helper libraries and set your normal OpenAI security credentials for di OpenAI Service wey you don create  
3. Choose one model for your work  
4. Create simple prompt for di model  
5. Submit your request to di model API!  


### 1. Install OpenAI


```python
%pip install openai python-dotenv
```

### 2. Import helper libraries an start credentials


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Finding di correct model  
Models like GPT-4o and GPT-4o mini fit sabi and create natural language, dem dey available on top OpenAI platform with different level of power and speed wey fit different work. 


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

## 4. Prompt Design  

"Di magic of big language models na sey as dem dey train am to reduce dis kind mistake for plenty plenty text, di models go dey learn tins wey go help dem do di prediction well well. For example, dem go sabi tins like"(1):

* how to spell
* how grammar dey work
* how to paraphrase
* how to answer questions
* how to hold conversation
* how to write for plenty languages
* how to code
* etc.

#### How to control big language model  
"For all di tins wey you fit put for big language model, di most important na di text prompt(1).

Big language models fit use different ways to prompt am make e give output:

Instruction: Tell di model wetin you want
Completion: Make di model finish di beginning of wetin you want
Demonstration: Show di model wetin you want, e fit be either:
Small examples for di prompt
Plenty hundred or thousand examples for fine-tuning training dataset"



#### Dem get three main rules for making prompts:

**Show and tell**. Make e clear wetin you want by instruction, examples, or both. If you want di model arrange list of tins for alphabetical order or classify paragraph by how e feel, show am sey na di tin you want.

**Give better data**. If you dey try build classifier or make model follow pattern, make sure say you get enough examples. Make you check your examples well — di model fit see small small spelling mistake and still give answer, but e fit think say you do am on purpose and e fit change di answer.

**Check your settings.** Temperature and top_p settings dey control how sure di model go be when e dey give answer. If you dey find answer wey only get one correct, you go set dem low. If you want different kain answers, you fit set dem high. Di main mistake wey pipo dey make with these settings na to think say na "cleverness" or "creativity" control dem be.


Source: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Submit!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

### If you call am again, how di results go compare?


```python

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
