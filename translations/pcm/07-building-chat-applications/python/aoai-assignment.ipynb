### 2. Import helper libraries and start credentials


```python
import os
from openai import OpenAI
import numpy as np
from dotenv import load_dotenv
load_dotenv()

#validate data inside .env file

# The Responses API is served from the Azure OpenAI (Microsoft Foundry) v1 endpoint,
# so we point the OpenAI client at <your-endpoint>/openai/v1/ (no api_version needed).
endpoint = os.environ['AZURE_OPENAI_ENDPOINT']
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{endpoint.rstrip('/')}/openai/v1/",
  )
```

### 3. Finding di correct model

```python
# Select the deployment name configured in your .env file
model = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

#### How to control large language model  
"For all di input wey large language model fit get, di text prompt na di one wey get di biggest influence(1).

Large language models fit get prompt to produce output for some ways:

- Instruction: Tell di model wetin you want
- Completion: Make di model finish di beginning of wetin you want
- Demonstration: Show di model wetin you want, with either:
- Small examples inside di prompt
- Plenty hundreds or thousands examples for fine-tuning training dataset



#### Di basic guideline for making prompts na dis three ones:

**Show and tell**. Make am clear wetin you want either through instructions, examples, or both. If you want di model to rank list of tins for alphabetical order or to classify paragraph by sentiment, show am say na dat thing you want.

**Provide better data**. If you dey try build classifier or make model follow pattern, make sure sey enough examples dey. Make you proofread your examples — model fit sabi well to detect simple spelling mistake dem and still give you answer, but e fit also think say you do am on purpose and e fit affect how e answer.

**Check your settings.** Di temperature and top_p settings na wetin control how deterministic model go be to give answer. If you dey ask for answer wey get only one correct one, you go wan set dem low. If you dey find more variety answer, you fit set dem high. Di number one mistake wey people dey do about dis settings na to think sey na "cleverness" or "creativity" controls be dat.

### 5. Submit!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Embeddings
Dis section go show how to find embeddings, and how to find similarities between words, sentences, and documents. To fit run the notebooks wey dey below, you go need to deploy model wey use `text-embedding-ada-002` as base model and set im deployment name for inside .env file, using the `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` variable.


### Model Taxonomy - How to sabi which embedding model to use

**Model taxonomy**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (dis na di embeddings family)  
{capability} --> ada             (all di oda embedding models go retire for 2024)  
{input-type} --> n/a             (dis one na only for search models)  
{identifier} --> 002             (version 002)  

model = 'text-embedding-ada-002'


  > [!NOTE] Di next step no too necessary if you dey run dis notebook for Codespaces or inside Devcontainer


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
