#### How to control big language model  
"Of all di inputs wey big language model get, di most important na di text prompt(1).

Big language models fit get prompt to produce output for few ways:

- Instruction: Tell di model wetin you want
- Completion: Make di model complete di beginning of wetin you want
- Demonstration: Show di model wetin you want, with either:
- Small examples for di prompt
- Plenty hundreds or thousands examples for fine-tuning training dataset



#### Dem get three basic guidelines to create prompts:

**Show and tell**. Make am clear wetin you want either through instructions, examples, or mix of both. If you want di model to rank list of items for alphabetical order or to classify paragraph by sentiment, show am say na wetin you want.

**Provide quality data**. If you dey try build classifier or make di model follow pattern, make sure say enough examples dey. Make sure say you proofread your examples — di model dey usually smart enough to see basic spelling mistakes and give you response, but e fit also think say na intentional and e fit affect di response.

**Check your settings.** Di temperature and top_p settings dey control how deterministic di model go be for generating response. If you dey ask am for response wey get only one correct answer, then you go want set dem lower. If you dey find more different responses, then you fit set dem higher. Di number one mistake wey people dey make with these settings na to think say na "cleverness" or "creativity" controls.

### 5. Submit!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.chat.completions.create(
  model=model,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```

## Embeddings
Dis section go show how to get embeddings, and find similarities between words, sentences, and documents. To fit run di following notebooks, you need to deploy model wey dey use `text-embedding-ada-002` as base model and set im deployment name inside .env file, using di `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` variable.


### Model Taxonomy - How to choose embedding model

**Model taxonomy**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (embedding family)  
{capability} --> ada             (all di oda embedding models go retire for 2024)  
{input-type} --> n/a             (dem only dey specify am for search models)  
{identifier} --> 002             (version 002)  

model = 'text-embedding-ada-002'


  > [!NOTE] Di next step no too necessary if you dey run dis notebook for Codespaces or inside one Devcontainer


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
