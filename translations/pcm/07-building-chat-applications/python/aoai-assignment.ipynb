## 4. Prompt Design  

"Di magic wey dey large language models be say as dem dey train dem to reduce di prediction error for plenty text, di models go dey learn concepts wey go help dem for dis predictions. For example, dem dey learn concepts like"(1):

#### How to control large language model  
"From all di inputs wey large language model dey get, na di text prompt be di one wey dey influence am pass(1).

Large language models fit dey prompted to produce output in some ways:

- Instruction: Tell di model wetin you want
- Completion: Make di model complete di beginning of wetin you want
- Demonstration: Show di model wetin you want, with either:
- Few examples for di prompt
- Many hundreds or thousands of examples for fine-tuning training dataset



#### Three basic guidelines to create prompts:

**Show and tell**. Make am clear wetin you want either through instructions, examples, or mix di two. If you want di model to arrange list of items for alphabetical order or to classify one paragraph by sentiment, show am say na wetin you want.

**Provide quality data**. If you dey try build one classifier or make di model follow one pattern, make sure say you get enough examples. Make sure say you proofread your examples — di model dey usually smart enough to see through basic spelling mistakes and give you response, but e fit also assume say na intentional and e fit affect di response.

**Check your settings.** Di temperature and top_p settings dey control how di model go dey sure for di response wey e dey generate. If you dey ask am for response wey get only one correct answer, you go wan set dis ones low. If you dey look for more diverse responses, you fit wan set dem high. Di number one mistake wey people dey make with dis settings na to assume say dem be "cleverness" or "creativity" controls.

### 5. Submit!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.chat.completions.create(
  model=model,
  messages = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},])

response.choices[0].message.content
```

## Embeddings
Dis section go show how you fit take get embeddings, and find similarities between words, sentences, and documents. To fit run di notebooks wey dey follow, you go need deploy one model wey dey use `text-embedding-ada-002` as base model and put im deployment name inside .env file, using di `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` variable.


### Model Taxonomy - How to choose embedding model

**Model taxonomy**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (embeddings family)  
{capability} --> ada             (all di oda embedding models go retire for 2024)  
{input-type} --> n/a             (na only search models dem dey specify dis one)  
{identifier} --> 002             (version 002)  

model = 'text-embedding-ada-002'


> [!NOTE] Dis step no dey necessary if you dey run dis notebook for Codespaces or inside Devcontainer


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
