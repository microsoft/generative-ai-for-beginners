## Llama 3.1 

Wif 405 Billion Parameters, Llama 3.1 dey for open source LLM category. 

Dis mode na upgrade to di earlier release Llama 3 by offering: 

- Bigger context window - 128k tokens vs 8k tokens 
- Bigger Max Output Tokens - 4096 vs 2048 
- Better Multilingual Support - because e get more training tokens 

Dis one allow Llama 3.1 make e fit handle more complex use cases wen you dey build GenAI applications including: 
- Native Function Calling - ability to call external tools and functions outside di LLM workflow
- Better RAG Performance - because e get bigger context window 
- Synthetic Data Generation - ability to create better data for tasks like fine-tuning 



### Native Function Calling 

Llama 3.1 don fine-tune to fit make e better for sabi how to call function or tool dem. E still get two built-in tool wey model fit sabi say e suppose use dem based on wetin user talk for prompt. Dem tools dem na:

- **Brave Search** - Fit use am take find up-to-date info like weather by to do web search
- **Wolfram Alpha** - Fit use am do complex maths calculation, so you no need write your own functions.

You fit also create your own custom tool wey LLM fit call.

For the code example wey dey down:

- We dey define the tools wey dey available (brave_search, wolfram_alpha) for the system prompt.
- Send user prompt wey dey ask about weather for one city.
- The LLM go respond with tool call to Brave Search tool wey go look like this `<|python_tag|>brave_search.call(query="Stockholm weather")`

*Note: This example only dey make the tool call, if you want get the results, you go need create free account for the Brave API page and define the function itself` 


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### Llama 3.2 

Even though na LLM e be, one limitation wey Llama 3.1 get na multimodality. Dat one mean sey e fit use different kain input like pictures as prompts and still fit give response. Dis kain skill na one of di main features wey Llama 3.2 get. Dem features still include: 

- Multimodality - e get di power to sabi text and image prompts 
- Small to Medium size variations (11B and 90B) - dis one dey give flexible deployment options, 
- Text-only variations (1B and 3B) - dis one make di model fit run for edge / mobile devices and e dey give low latency 

Di multimodal support na one big step for di world of open source models. Di code example wey dey below go take both picture and text prompt to do analysis of di image from Llama 3.2 90B. 

### Multimodal Support with Llama 3.2


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
