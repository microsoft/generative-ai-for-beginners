## Generate text

To generate text, you go use `responses.create` method. See example:

```python

```

For code wey dey above, we create response object and pass model wey we wan use plus the prompt. Then we print the generated text.

### Chat responses

So far you don see how we fit generate text from one prompt. The Responses API fit as well for chatbots - you go give am list of messages as the `input` to build up conversation. Example here:


1. Find the `prompt` variable and change the code to the one below:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
