## Generate text

How you go generate text be say use `responses.create` method. Example like dis:

```python

```

For code wey pass, we create response object and pass model wey we wan use plus prompt. Then we print generated text.

### Chat responses

So far, you don see how to generate text from one prompt. Responses API dey well for chatbots too - you dey provide list of messages as `input` wey build conversation. Example dey like dis:

```python

```

- **Talk about token length**. We suppose check how many tokens we need make we generate the text we want. Tokens dey cost money, so where e possible, make we try save tokens. For example, fit we talk di prompt in way we go fit use less tokens?
