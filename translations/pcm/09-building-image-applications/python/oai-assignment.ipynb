## Wetin be image generation?

Image generation models dey create images from text prompt. Modern models like `gpt-image` dey learn how text relate to images during training, then dem dey convert random noise to image wey match your prompt step by step.

Two well-known image model family na:

- **`gpt-image` (OpenAI)** — na di current generation wey we dey use for dis lesson. E fit do text-to-image generation and image editing (inpainting with mask).
- **Midjourney** — popular third-party model wey get im own service and im workflow dey happen for Discord.

> Di old OpenAI image models — **DALL·E 2** and **DALL·E 3** — na legacy. DALL·E 3 no dey available for new deployments again, and features like `create_variation` only dey for DALL·E 2. Make you use `gpt-image` models for new applications.

> **Important:** `gpt-image` models dey return di generated image as **base64** (`b64_json`), no be as URL. Your code go decode di base64 string to bytes and save am — no image URL dey for download.


## How to build your first image generation app

So wetin you need to build image generation app? You go need dis libraries:

- **python-dotenv**, e beta make you use dis library to keep your secrets inside * .env * file wey dey separate from code.
- **openai**, dis library na wetin you go use take dey interact with OpenAI API.
- **pillow**, to work with images for Python.


1. Make one file *.env* with dis kind content:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. Gather the libraries wey dey up for a file wey dem go call *requirements.txt* like dis:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. Next, create virtual environment and install the libraries:

> [!NOTE]
> For Windows, use dis commands dem below to create and activate your virtual environment:

1. Add dis code below for file wey dem call *app.py*:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # import dotenv
    dotenv.load_dotenv()

    # Create OpenAI object (e dey read OPENAI_API_KEY from your .env)
    client = openai.OpenAI()


    try:
        # Make image wit di image generation API
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Put your prompt text here
            size='1024x1024',
            n=1
        )
        # Set di folder wey go hold di image
        image_dir = os.path.join(os.curdir, 'images')

        # If di folder no dey, make am
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Start di image path (make sure say di file na png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image models dey return image as base64 (b64_json), no be URL
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Show di image for di default image viewer
        image = Image.open(image_path)
        image.show()

    # catch exceptions
    except openai.BadRequestError as err:
        print(err)

    ```

### More details about how to generate di image

Make we look di parameters wey dey `images.generate`:

- **model**, na di image model wey you go use (for example `gpt-image-1`).
- **prompt**, na di text prompt wey dem use generate di image. Here na "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, na di size of di image wey dem generate (for example `1024x1024`, `1536x1024`, `1024x1536`, or `"auto"`).
- **n**, na di number of images wey dem generate. Here we generate one.

> Image models no dey use `temperature` parameter — dat one na text-generation control. If you want different kain, call di API again; if you want less different kain, make your prompt more specific.

## Mo capabilities for image generation

You don see how to generate image with small Python code dem. `gpt-image` models fit still **edit** image wey already dey. If you give image, plus optional **mask** (wey go mark di area wey you wan change), and prompt, you fit change part of image — like add hat for our bunny.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# edits dem sef go return as base64
edited_image = base64.b64decode(response.data[0].b64_json)
```

Di base image get only di rabbit; di final image get hat for di rabbit.
