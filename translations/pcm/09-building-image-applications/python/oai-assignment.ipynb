## Wetin be DALL-E and Midjourney? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) and [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) na two of di popular image generation models, dem dey allow you use prompts to generate pictures.

### DALL-E

Make we start with DALL-E, na Generative AI model wey dey generate pictures from text description. 

> [DALL-E na combination of two models, CLIP and diffused attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **CLIP**, na model wey dey generate embeddings, wey be numerical way wey data dey represent, from pictures and text.  

- **Diffused attention**, na model wey dey generate pictures from the embeddings. DALL-E train with plenty pictures and text and fit use am generate pictures come from text description. For example, DALL-E fit generate pictures of cat wey get hat, or dog wey get mohawk. 

### Midjourney
 
Midjourney dey work almost like DALL-E, e dey generate pictures from text prompts. Midjourney fit also generate pictures with prompts like “cat in hat”, or “dog with mohawk”.

## How DALL-E and Midjourney Dey Work 

First, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E na Generative AI model based on transformer architecture with *autoregressive transformer*.  

*Autoregressive transformer* na how model dey generate pictures from text description, e dey generate one pixel every time, then e use the pixels wey e generate before come generate next pixel. E go pass through many layers for neural network till picture complete.

## Di way to build your first image generation app

So wetin you need to build app wey go make image? You go need dis kin libraries:

1. Put dis code for file wey dem call *app.py*:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # import dotenv
    dotenv.load_dotenv()

    # Make OpenAI object
    client = openai.OpenAI()


    try:
        # Make one picture by using di image generation API
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Put your prompt text here
            size='1024x1024',
            n=1
        )
        # Set di folder wey go hold di saved picture
        image_dir = os.path.join(os.curdir, 'images')

        # If di folder no dey, make am
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Start di image path (make sure say di file type be png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # Collect di generated picture
        print(generation_response)

        image_url = generation_response.data[0].url # comot di image URL from di response
        generated_image = requests.get(image_url).content  # download di picture
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Show di picture for di default image viewer
        image = Image.open(image_path)
        image.show()

    # catch exceptions
    except openai.BadRequestError as err:
        print(err)

    ```

## Additional capabilities of image generation

You don see so far how we fit generate image wit small lines for Python. But e get more tins we fit do wit images.

You fit also do dis kain tins:

- **Perform edits**. If you get one existing image, plus mask and prompt, you fit change di image. For example, you fit add somtin for one part of di image. Imagine our bunny image, you fit put hat for di bunny. How you go do am be say you go give di image, one mask (wey show which part you wan change) plus di text prompt wey talk wetin to do.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    Di base image go get only di rabbit but di final image go get di hat for di rabbit.
    
- **Create variations**. Di idea be say you go take one existing image and ask make dem create variations. To create variation, you go give one image and one text prompt and code like dis:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
