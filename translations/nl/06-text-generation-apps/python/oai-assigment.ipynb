## Wat is een tekstgeneratie-app?

Normaal gesproken heeft een app een soort interface zoals de volgende:

- Op commando's gebaseerd. Console-apps zijn typische apps waar je een commando typt en het een taak uitvoert. Bijvoorbeeld, `git` is een op commando's gebaseerde app.
- Gebruikersinterface (UI). Sommige apps hebben grafische gebruikersinterfaces (GUI's) waarbij je op knoppen klikt, tekst invoert, opties selecteert en meer.

### Console- en UI-apps zijn beperkt

Vergelijk het met een op commando's gebaseerde app waar je een commando typt:

- **Het is beperkt**. Je kunt niet zomaar elk commando typen, alleen degene die de app ondersteunt.
- **Taalgebonden**. Sommige apps ondersteunen veel talen, maar standaard is de app gebouwd voor een specifieke taal, zelfs als je extra taalsupport kunt toevoegen.

### Voordelen van tekstgeneratie-apps

Hoe verschilt een tekstgeneratie-app dan?

In een tekstgeneratie-app heb je meer flexibiliteit, je bent niet beperkt tot een set commando's of een specifieke invoertaal. In plaats daarvan kun je natuurlijke taal gebruiken om met de app te communiceren. Een ander voordeel is dat je interactie hebt met een databron die getraind is op een enorme hoeveelheid informatie, terwijl een traditionele app beperkt kan zijn tot wat er in een database staat.

### Wat kan ik bouwen met een tekstgeneratie-app?

Er zijn veel dingen die je kunt bouwen. Bijvoorbeeld:

- **Een chatbot**. Een chatbot die vragen beantwoordt over onderwerpen zoals jouw bedrijf en zijn producten kan een goede match zijn.
- **Helper**. LLM's zijn goed in dingen zoals het samenvatten van tekst, het verkrijgen van inzichten uit tekst, het produceren van teksten zoals cv's en meer.
- **Code-assistent**. Afhankelijk van het gebruikte taalmodel kun je een code-assistent bouwen die helpt bij het schrijven van code. Bijvoorbeeld, je kunt een product zoals GitHub Copilot gebruiken evenals ChatGPT om je te helpen met coderen.

## Genereer tekst

De manier om tekst te genereren is door de methode `responses.create` te gebruiken. Hier is een voorbeeld:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

In de bovenstaande code maken we een response-object en geven we het model dat we willen gebruiken en de prompt aan. Daarna printen we de gegenereerde tekst.

### Chat-antwoorden

Tot nu toe heb je gezien hoe we vanuit een enkele prompt tekst kunnen genereren. De Responses API is ook uitstekend geschikt voor chatbots - je geeft een lijst met berichten als `input` om een gesprek op te bouwen. Hier is een voorbeeld van hoe je het gebruikt:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## Oefening - bouw een receptenmaker

Nu we een scenario hebben doorgenomen, laten we code schrijven die overeenkomt met het getoonde scenario. Volg hiervoor de volgende stappen:

1. Gebruik het bestaande *app.py* bestand als uitgangspunt
1. Zoek de `prompt` variabele op en vervang de code ervan door het volgende:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Geweldig, laten we eens kijken hoe we de zaken kunnen verbeteren. Om dingen te verbeteren, willen we ervoor zorgen dat de code flexibel is, zodat ingrediënten en het aantal recepten kunnen worden verbeterd en gewijzigd.


1. Laten we de code op de volgende manier aanpassen:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Verbeteren door filter en boodschappenlijst toe te voegen

We hebben nu een werkende app die recepten kan produceren en flexibel is omdat het afhangt van input van de gebruiker, zowel qua aantal recepten als gebruikte ingrediënten.

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"

- **Een boodschappenlijst maken**. We willen een boodschappenlijst kunnen maken, rekening houdend met wat we al in huis hebben.

    Voor deze functionaliteit kunnen we alles in één prompt proberen op te lossen, of we splitsen het op in twee prompts. Laten we de laatste aanpak proberen. Hier stellen we voor een extra prompt toe te voegen, maar daarvoor moeten we het resultaat van de eerste prompt als context toevoegen aan de tweede prompt.

    Zoek het gedeelte in de code waar het resultaat van de eerste prompt wordt afgedrukt en voeg onderstaande code toe eronder:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print antwoord
    print("Shopping list:")
    print(response.output_text)
    ```

    Let op het volgende:

    - We bouwen een nieuwe prompt door het resultaat van de eerste prompt toe te voegen aan de nieuwe prompt:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We doen een nieuw verzoek, rekening houdend met het aantal tokens dat we bij de eerste prompt vroegen, dus deze keer stellen we `max_output_tokens` in op 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Als we deze code uitvoeren, krijgen we nu de volgende output:

- **Een woord over tokenlengte**. We moeten overwegen hoeveel tokens we nodig hebben om de tekst te genereren die we willen. Tokens kosten geld, dus waar mogelijk moeten we economisch zijn met het aantal tokens dat we gebruiken. Kunnen we bijvoorbeeld de prompt anders formuleren om minder tokens te gebruiken?

   Om het aantal gebruikte tokens aan te passen, kun je de parameter `max_output_tokens` gebruiken. Bijvoorbeeld, als je 100 tokens wilt gebruiken, doe je:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenteren met temperatuur**. Temperatuur hebben we tot nu toe niet genoemd, maar het is een belangrijke factor voor hoe ons programma presteert. Hoe hoger de temperatuur, hoe willekeuriger de output. Hoe lager de temperatuur, hoe voorspelbaarder de output. Bedenk of je variatie in je output wilt of niet.

   Om de temperatuur te wijzigen, gebruik je de parameter `temperature`. Bijvoorbeeld, als je een temperatuur van 0.5 wilt instellen, doe je:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Let op: hoe dichter bij 1.0, hoe gevarieerder de output.

- Bouw een "studiepartner". Deze app moet vragen over een onderwerp, bijvoorbeeld Python, kunnen beantwoorden. Je zou prompts kunnen hebben als "Wat is een bepaald onderwerp in Python?" of een prompt die vraagt om code te laten zien van een bepaald onderwerp, enzovoorts.
- Geschiedenis-bot: laat geschiedenis tot leven komen, instrueer de bot om een bepaald historisch figuur te spelen en stel hem vragen over zijn leven en tijd.
