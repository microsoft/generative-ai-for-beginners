## Wat is een tekstgeneratie-app?

Normaal gesproken als je een app bouwt, heeft die een soort interface zoals de volgende:

- Commando-gebaseerd. Console-apps zijn typische apps waarbij je een commando typt en die een taak uitvoert. Bijvoorbeeld, `git` is een commando-gebaseerde app.
- Gebruikersinterface (UI). Sommige apps hebben grafische gebruikersinterfaces (GUI's) waar je op knoppen klikt, tekst invoert, opties selecteert en meer.

### Console- en UI-apps zijn beperkt

Vergelijk het met een commando-gebaseerde app waarbij je een commando typt:

- **Het is beperkt**. Je kunt niet zomaar elk commando typen, alleen die welke de app ondersteunt.
- **Taal specifiek**. Sommige apps ondersteunen veel talen, maar standaard is de app gebouwd voor een specifieke taal, ook al kun je meer taalondersteuning toevoegen.

### Voordelen van tekstgeneratie-apps

Dus hoe is een tekstgeneratie-app anders?

In een tekstgeneratie-app heb je meer flexibiliteit, je bent niet beperkt tot een set commando's of een specifieke invoertaal. In plaats daarvan kun je natuurlijke taal gebruiken om met de app te communiceren. Een ander voordeel is dat je al communiceert met een gegevensbron die getraind is op een enorme verzameling informatie, terwijl een traditionele app mogelijk beperkt is tot wat er in een database staat.

### Wat kan ik bouwen met een tekstgeneratie-app?

Er zijn veel dingen die je kunt bouwen. Bijvoorbeeld:

- **Een chatbot**. Een chatbot die vragen over onderwerpen beantwoordt, zoals over jouw bedrijf en zijn producten, kan een goede match zijn.
- **Helper**. LLM's zijn geweldig in dingen zoals het samenvatten van tekst, inzichten halen uit tekst, tekst produceren zoals cv's, en meer.
- **Code assistent**. Afhankelijk van het taalmodel dat je gebruikt, kun je een code-assistent bouwen die je helpt code te schrijven. Bijvoorbeeld, je kunt een product zoals GitHub Copilot gebruiken evenals ChatGPT om je te helpen code te schrijven.

## Genereer tekst

De manier om tekst te genereren is door de `responses.create` methode te gebruiken. Hier is een voorbeeld:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

In bovenstaande code maken we een response-object aan en geven we het model dat we willen gebruiken en de prompt door. Daarna printen we de gegenereerde tekst.

### Chat-antwoorden

Tot nu toe heb je gezien hoe we tekst kunnen genereren vanuit een enkele prompt. De Responses API is ook zeer geschikt voor chatbots - je geeft een lijst met berichten door als `input` om een gesprek op te bouwen. Hier is een voorbeeld van het gebruik:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)

```

## Oefening - bouw een receptgenerator

Nu we een scenario hebben gespeeld, laten we code schrijven die overeenkomt met het getoonde scenario. Volg hiervoor de volgende stappen:

1. Gebruik het bestaande *app.py* bestand als uitgangspunt
1. Zoek de variabele `prompt` en verander de code ervan naar het volgende:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. Laten we de code als volgt wijzigen:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Verbeteren door filter en boodschappenlijst toe te voegen

- **Een boodschappenlijst produceren**. We willen een boodschappenlijst maken, rekening houdend met wat we al thuis hebben.

    Voor deze functionaliteit kunnen we ofwel alles in één prompt proberen op te lossen of het in twee prompts splitsen. Laten we de laatste aanpak proberen. Hier stellen we voor een extra prompt toe te voegen, maar daarvoor moeten we de uitkomst van de eerste prompt als context aan de tweede prompt toevoegen.

    Zoek het deel in de code dat het resultaat van de eerste prompt afdrukt en voeg onderstaande code toe:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print antwoord
    print("Shopping list:")
    print(response.output_text)
    ```

    Let op het volgende:

    - We construeren een nieuwe prompt door het resultaat van de eerste prompt toe te voegen aan de nieuwe prompt:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We maken een nieuw verzoek, maar rekening houdend met het aantal tokens dat we in de eerste prompt vroegen, dus deze keer zeggen we dat `max_output_tokens` 1200 is.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```

- **Een woord over tokenlengte**. We moeten overwegen hoeveel tokens we nodig hebben om de tekst te genereren die we willen. Tokens kosten geld, dus waar mogelijk moeten we economisch proberen te zijn met het aantal tokens dat we gebruiken. Kunnen we bijvoorbeeld de prompt zo formuleren dat we minder tokens gebruiken?

   Om het gebruikte aantal tokens te wijzigen kun je de parameter `max_output_tokens` gebruiken. Bijvoorbeeld, als je 100 tokens wilt gebruiken zou je doen:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenteren met temperatuur**. Temperatuur is iets wat we tot nu toe niet benoemd hebben maar is een belangrijke context voor hoe ons programma presteert. Hoe hoger de temperatuurwaarde, hoe willekeuriger de output zal zijn. Omgekeerd, hoe lager de temperatuurwaarde, hoe voorspelbaarder de output zal zijn. Bedenk of je variatie in je output wilt of niet.

   Om de temperatuur aan te passen kun je de parameter `temperature` gebruiken. Bijvoorbeeld, als je een temperatuur van 0.5 wilt gebruiken, doe je het volgende:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Opmerking: hoe dichter bij 1.0, hoe gevarieerder de output.

- Pas de receptgenerator-app aan om deze verder te verbeteren. Experimenteer met temperatuurwaarden en de prompts om te zien wat je kunt bedenken.
- Bouw een "studiegenoot". Deze app moet in staat zijn vragen over een onderwerp te beantwoorden, bijvoorbeeld Python, je zou prompts kunnen hebben als "Wat is een bepaald onderwerp in Python?", of een prompt die zegt, laat me code zien voor een bepaald onderwerp etc.
- Geschiedenisbos, laat geschiedenis tot leven komen, instrueer de bot om een bepaald historisch personage te spelen en stel het vragen over diens leven en tijd.

## Oplossing

### Studienoot

- "Je bent een expert in de Python-taal

    Stel een beginnersles voor Python voor in het volgende formaat:
    
    Formaat:
    - concepten:
    - korte uitleg van de les:
    - oefening in code met oplossingen"

Bovenstaande is een startprompt, kijk hoe je die kunt gebruiken en aanpassen naar je eigen smaak.

### Geschiedenisbos

Hier zijn enkele prompts die je zou kunnen gebruiken:

- "Je bent Abe Lincoln, vertel in 3 zinnen over jezelf, en reageer met grammatica en woorden zoals Abe dat zou doen"
- "Je bent Abe Lincoln, reageer met grammatica en woorden zoals Abe dat zou doen:

   Vertel me over je grootste prestaties, in 300 woorden:"

## Kennischeck

1. Het bepaalt hoe willekeurig de output is.
1. Het bepaalt hoe groot de respons is.
1. Het bepaalt hoeveel tokens er gebruikt worden.

A: 1

Wat is een goede manier om geheimen zoals API-sleutels op te slaan?

1. In de code.
1. In een bestand.
