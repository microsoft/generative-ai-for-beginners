```python
import os
from openai import OpenAI

API_KEY = os.getenv("AZURE_OPENAI_API_KEY","").strip()
assert API_KEY, "ERROR: Azure OpenAI Key is missing"

RESOURCE_ENDPOINT = os.getenv("AZURE_OPENAI_ENDPOINT","").strip()
assert RESOURCE_ENDPOINT, "ERROR: Azure OpenAI Endpoint is missing"
assert "openai.azure.com" in RESOURCE_ENDPOINT.lower(), "ERROR: Azure OpenAI Endpoint should be in the form: \n\n\t<your unique endpoint identifier>.openai.azure.com"

deployment = os.getenv("AZURE_OPENAI_DEPLOYMENT", "").strip()
assert deployment, "ERROR: Azure OpenAI Deployment name is missing"

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
# for the Responses API - no api_version needed.
client = OpenAI(
    api_key=API_KEY,
    base_url=f"{RESOURCE_ENDPOINT.rstrip('/')}/openai/v1/",
)
```

```python
# Create your first prompt
text_prompt = " My foot hurts, what can be wrong?"

response = client.responses.create(
  model=deployment,
  input = [
      {"role":"system", "content":"I'm a doctor, specialist on surgery"},
      {"role":"user","content":text_prompt},],
  store=False,)


response.output_text
```
