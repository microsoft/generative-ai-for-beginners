## テキスト生成アプリとは？

通常アプリを作ると、次のようなインターフェイスが付きものです:

- コマンドベース。コンソールアプリは典型的で、コマンドをタイプするとそれが処理されるタイプのアプリです。例えば `git` はコマンドベースのアプリです。
- ユーザーインターフェイス（UI）。グラフィカルユーザーインターフェイス（GUI）を持ち、ボタンをクリックしたり、テキストを入力したり、オプションを選択したりするアプリもあります。

### コンソール・UIアプリは制限がある

コマンドをタイプするコマンドベースのアプリに比べると:

- <strong>制限されている</strong>。使えるコマンドはアプリがサポートするものに限られ、それ以外は使えない。
- <strong>言語依存</strong>。多くの言語がサポートできる場合もありますが、通常は特定の言語用に作られており、追加の言語サポートはオプションでしかない。

### テキスト生成アプリの利点

ではテキスト生成アプリはどう違うのでしょうか？

テキスト生成アプリの場合、より柔軟に使え、限られたコマンドや特定の入力言語に縛られません。代わりに自然言語を使ってアプリと対話できます。さらに、すでに広範な情報コーパスで訓練されたデータソースとやり取りしているため、従来のデータベースに限定されたアプリよりも幅広い情報を扱えます。

### テキスト生成アプリで何が作れる？

さまざまなものが作れます。例えば:

- <strong>チャットボット</strong>。会社や製品に関する質問に答えるチャットボットなどが適しています。
- <strong>ヘルパー</strong>。LLMはテキストの要約やインサイト抽出、履歴書の作成などテキスト作成に優れています。
- <strong>コードアシスタント</strong>。使うモデルによっては、コード作成を支援するコードアシスタントを作れます。GitHub Copilot や ChatGPT などを利用してコードを書く手助けも可能です。

## 演習 - レシピジェネレーターを作成する

シナリオを実行したので、示されたシナリオに合うコードを書いてみましょう。そのために、以下の手順に従ってください：

1. 既存のファイルを出発点として使用する
1. `prompt` 変数を作成し、サンプルコードを以下のように変更する：


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
