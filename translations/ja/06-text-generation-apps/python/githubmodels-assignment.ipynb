## テキスト生成アプリとは？

通常アプリを作るとき、何らかのインターフェースがあります。たとえば：

- コマンドベース。コンソールアプリは典型的なアプリで、コマンドを入力すると特定のタスクを実行します。たとえば`git`はコマンドベースのアプリです。
- ユーザーインターフェース（UI）。GUIを持つアプリではボタンをクリックしたり、テキスト入力やオプション選択などができます。

### コンソールとUIアプリの制限

コマンドベースのアプリに入力するコマンドと比べてみてください：

- <strong>制限がある</strong>。入力できるコマンドは、アプリがサポートしているものだけです。
- <strong>言語特有</strong>。いくつかのアプリは多言語対応ですが、デフォルトでは特定の言語向けに作られていて、後から言語サポートを追加できます。

### テキスト生成アプリの利点

テキスト生成アプリはどう違うでしょう？

テキスト生成アプリでは、より柔軟に自然言語を使って操作でき、特定のコマンドセットや入力言語に限定されません。また、既に膨大な情報で学習されたデータソースと対話するため、従来のデータベースに制限されていたアプリより幅広い対応が可能です。

### テキスト生成アプリで何が作れる？

さまざまなものが作れます。たとえば：

- <strong>チャットボット</strong>。会社や製品について質問に答えるチャットボットは良い例です。
- <strong>ヘルパー</strong>。LLMは文章の要約や洞察の取得、履歴書などの文章生成に優れています。
- <strong>コードアシスタント</strong>。使用する言語モデルにより、コード作成を助けるアシスタントが作れます。たとえばGitHub CopilotやChatGPTがコード作成を支援します。

## 演習 - レシピジェネレーターを作る

シナリオを実行したので、示したシナリオに合うコードを書いてみましょう。そのために、以下の手順に従ってください:

1. 既存のファイルを出発点として使う
1. `prompt` 変数を作成し、サンプルコードを以下のように変更する


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
