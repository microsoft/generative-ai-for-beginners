## Czym jest aplikacja generująca tekst?

Zazwyczaj, gdy tworzysz aplikację, ma ona jakiś interfejs, na przykład:

- Bazujący na poleceniach. Aplikacje konsolowe to typowe aplikacje, w których wpisujesz polecenie, a ona wykonuje zadanie. Na przykład `git` to aplikacja bazująca na poleceniach.
- Interfejs użytkownika (UI). Niektóre aplikacje mają graficzne interfejsy użytkownika (GUI), w których klikasz przyciski, wpisujesz tekst, wybierasz opcje i więcej.

### Konsolowe i GUI są ograniczone

Porównaj to z aplikacją bazującą na poleceniach, gdzie wpisujesz polecenie:

- **Jest to ograniczone**. Nie możesz po prostu wpisać dowolnego polecenia, tylko te, które aplikacja obsługuje.
- **Język specyficzny**. Niektóre aplikacje obsługują wiele języków, ale domyślnie aplikacja jest stworzona dla konkretnego języka, nawet jeśli możesz dodać wsparcie dla innych języków.

### Zalety aplikacji generujących tekst

Czym więc różni się aplikacja generująca tekst?

W aplikacji generującej tekst masz większą elastyczność, nie jesteś ograniczony do zestawu poleceń ani konkretnego języka wejściowego. Zamiast tego możesz używać naturalnego języka do interakcji z aplikacją. Kolejną zaletą jest to, że korzystasz z źródła danych, które zostało przeszkolone na ogromnym zbiorze informacji, podczas gdy tradycyjna aplikacja może być ograniczona do danych znajdujących się w bazie.

### Co mogę zbudować za pomocą aplikacji generującej tekst?

Możesz stworzyć wiele rzeczy, na przykład:

- **Chatbota**. Chatbot odpowiadający na pytania na tematy związane z twoją firmą i jej produktami może być dobrym rozwiązaniem.
- **Pomocnika**. Modele LLM świetnie sprawdzają się w takich zadaniach jak streszczanie tekstu, wydobywanie wniosków, tworzenie tekstów takich jak CV i więcej.
- **Asystenta kodu**. W zależności od używanego modelu językowego możesz stworzyć asystenta kodu, który pomaga pisać kod. Na przykład można korzystać z produktów takich jak GitHub Copilot oraz ChatGPT, by pomagać przy pisaniu kodu.

## Ćwiczenie - stwórz generator przepisów

Teraz, gdy przećwiczyliśmy scenariusz, napiszmy kod odpowiadający zaprezentowanemu scenariuszowi. Aby to zrobić, wykonaj następujące kroki:

1. Użyj istniejącego pliku jako punktu wyjścia
1. Utwórz zmienną `prompt` i zmień przykładowy kod jak poniżej:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

### Ulepsz przez dodanie filtra i listy zakupów

Mamy teraz działającą aplikację zdolną do generowania przepisów i jest elastyczna, ponieważ opiera się na danych wejściowych od użytkownika, zarówno co do liczby przepisów, jak i używanych składników.

Aby ją dalej ulepszyć, chcemy dodać następujące elementy:

- **Filtrowanie składników**. Chcemy móc odfiltrować składniki, których nie lubimy lub na które jesteśmy uczuleni. W tym celu możemy edytować nasz istniejący prompt i dodać na końcu warunek filtra w taki sposób:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```
