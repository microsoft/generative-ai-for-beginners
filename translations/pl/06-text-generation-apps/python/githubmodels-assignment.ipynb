## Czym jest aplikacja do generowania tekstu?

Zwykle, gdy tworzysz aplikację, ma ona jakiś interfejs, na przykład taki:

- Oparta na poleceniach. Aplikacje konsolowe to typowe aplikacje, gdzie wpisujesz polecenie, a ono wykonuje zadanie. Na przykład `git` jest aplikacją opartą na poleceniach.
- Interfejs użytkownika (UI). Niektóre aplikacje mają graficzne interfejsy użytkownika (GUI), gdzie klikasz przyciski, wpisujesz tekst, wybierasz opcje i więcej.

### Aplikacje konsolowe i UI są ograniczone

Porównaj to do aplikacji opartej na poleceniach, gdzie wpisujesz polecenie:

- **Jest ograniczona**. Nie możesz wpisać dowolnego polecenia, tylko te, które aplikacja obsługuje.
- **Specyficzna dla języka**. Niektóre aplikacje obsługują wiele języków, ale domyślnie aplikacja jest zbudowana pod konkretny język, nawet jeśli możesz dodać większe wsparcie językowe.

### Zalety aplikacji do generowania tekstu

Czym różni się aplikacja do generowania tekstu?

W aplikacji do generowania tekstu masz więcej elastyczności, nie jesteś ograniczony do zestawu poleceń czy konkretnego języka wejściowego. Zamiast tego możesz korzystać z języka naturalnego do interakcji z aplikacją. Kolejną zaletą jest to, że ponieważ już korzystasz z źródła danych wytrenowanego na ogromnym korpusie informacji, podczas gdy tradycyjna aplikacja może być ograniczona do zawartości bazy danych.

### Co mogę zbudować z aplikacją do generowania tekstu?

Można zbudować wiele rzeczy. Na przykład:

- **Chatbot**. Chatbot odpowiadający na pytania o tematy, takie jak twoja firma i jej produkty, mógłby się dobrze sprawdzić.
- **Pomocnik**. LLM świetnie radzą sobie z takimi rzeczami jak podsumowywanie tekstu, uzyskiwanie informacji z tekstu, tworzenie tekstów takich jak CV i więcej.
- **Asystent kodowania**. W zależności od używanego modelu językowego można stworzyć asystenta kodowania, który pomaga pisać kod. Na przykład można użyć produktów takich jak GitHub Copilot oraz ChatGPT, aby pomóc pisać kod.

## Ćwiczenie - stwórz generator przepisów

Teraz, gdy przedstawiliśmy scenariusz, napiszmy kod odpowiadający zaprezentowanemu scenariuszowi. Aby to zrobić, wykonaj następujące kroki:

1. Użyj istniejącego pliku jako punktu wyjścia
1. Utwórz zmienną `prompt` i zmodyfikuj przykładowy kod w następujący sposób:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

### Ulepsz przez dodanie filtra i listy zakupów

Teraz mamy działającą aplikację zdolną do tworzenia przepisów, która jest elastyczna, ponieważ opiera się na danych wejściowych od użytkownika, zarówno liczbie przepisów, jak i składników.

Aby ją jeszcze bardziej ulepszyć, chcemy dodać następujące funkcje:

- **Odfiltruj składniki**. Chcemy mieć możliwość filtrowania składników, które nam się nie podobają lub na które mamy alergię. Aby dokonać tej zmiany, możemy edytować nasz istniejący prompt i dodać na jego końcu warunek filtru, w ten sposób:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Powyżej dodajemy `{filter}` na końcu promptu, a także przechwytujemy wartość filtru od użytkownika.

    Przykładowe uruchomienie programu może teraz wyglądać tak:
