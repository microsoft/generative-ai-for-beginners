## Czym jest aplikacja do generowania tekstu?

Zwykle, gdy tworzysz aplikację, ma ona jakiegoś rodzaju interfejs, na przykład:

- Opartą na poleceniach. Aplikacje konsolowe to typowe programy, gdzie wpisujesz polecenie, a one wykonują zadanie. Na przykład `git` jest aplikacją opartą na poleceniach.
- Interfejs użytkownika (UI). Niektóre aplikacje mają graficzne interfejsy użytkownika (GUI), gdzie klikasz przyciski, wprowadzasz tekst, wybierasz opcje i więcej.

### Aplikacje konsolowe i UI mają ograniczenia

Porównaj to do aplikacji opartej na poleceniach, gdzie wpisujesz komendę:

- **To jest ograniczone**. Nie możesz po prostu wpisać dowolnego polecenia, tylko te, które aplikacja obsługuje.
- **Język specyficzny**. Niektóre aplikacje obsługują wiele języków, ale domyślnie są zbudowane pod konkretny język, nawet jeśli możesz dodać wsparcie dla innych.

### Zalety aplikacji do generowania tekstu

Czym więc różni się aplikacja do generowania tekstu?

W aplikacji generującej tekst masz większą elastyczność, nie jesteś ograniczony do zestawu poleceń czy konkretnego języka wejściowego. Zamiast tego możesz używać naturalnego języka do interakcji z aplikacją. Kolejną korzyścią jest to, że wchodzisz w interakcję z źródłem danych, które zostało wytrenowane na ogromnym korpusie informacji, podczas gdy tradycyjna aplikacja może być ograniczona tym, co jest w bazie danych.

### Co mogę zbudować z aplikacją do generowania tekstu?

Możesz stworzyć wiele rzeczy. Na przykład:

- **Chatbota**. Chatbot odpowiadający na pytania dotyczące tematów, takich jak twoja firma i jej produkty, może być dobrym wyborem.
- **Pomocnika**. LLM świetnie sprawdzają się w zadaniach takich jak podsumowywanie tekstu, wydobywanie informacji z tekstu, tworzenie tekstów, takich jak CV i więcej.
- **Asystenta kodu**. W zależności od modelu językowego, którego używasz, możesz stworzyć asystenta kodu, który pomoże ci pisać kod. Na przykład możesz użyć produktów takich jak GitHub Copilot oraz ChatGPT, aby pomagały w pisaniu kodu.

## Generowanie tekstu

Sposobem na generowanie tekstu jest użycie metody `responses.create`. Oto przykład:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

W powyższym kodzie tworzymy obiekt odpowiedzi i przekazujemy model, którego chcemy użyć, oraz prompt. Następnie wypisujemy wygenerowany tekst.

### Odpowiedzi w chatcie

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)

    ```

    Zwróć uwagę na następujące rzeczy:

    - Tworzymy nowy prompt, dodając do niego wynik z pierwszego promptu:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Wysyłamy nowe zapytanie, ale bierzemy pod uwagę liczbę tokenów, o które poprosiliśmy w pierwszym promptcie, więc tym razem `max_output_tokens` wynosi 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Uruchamiając ten kod, otrzymujemy następujący wynik:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.
        
        -Apple and flour pancakes: 1 cup flour, 1/2 tsp baking powder, 1/2 tsp baking soda, 1/4 tsp salt, 1 tbsp sugar, 1 egg, 1 cup buttermilk or sour milk, 1/4 cup melted butter, 1 Granny Smith apple, peeled and grated
        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **Kilka słów o długości tokenów**. Powinniśmy rozważyć, ile tokenów potrzebujemy do wygenerowania pożądanego tekstu. Tokeny kosztują, więc tam, gdzie to możliwe, powinniśmy starać się oszczędzać ich liczbę. Na przykład, czy możemy sformułować prompt tak, by użyć mniej tokenów?

   Aby zmienić liczbę używanych tokenów, możesz użyć parametru `max_output_tokens`. Na przykład, jeśli chcesz użyć 100 tokenów, zrobisz tak:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Eksperymentowanie z temperaturą**. Temperatura to aspekt, o którym do tej pory nie wspomnieliśmy, ale jest ważnym parametrem wpływającym na działanie naszego programu. Im wyższa wartość temperatury, tym bardziej losowy jest wynik. Natomiast im niższa temperatura, tym bardziej przewidywalny wynik. Zastanów się, czy chcesz mieć zróżnicowanie w wynikach, czy nie.

   Aby zmienić temperaturę, możesz użyć parametru `temperature`. Na przykład, jeśli chcesz użyć temperatury 0.5, zrobisz tak:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Uwaga, im bliżej 1.0, tym bardziej zróżnicowany wynik.
