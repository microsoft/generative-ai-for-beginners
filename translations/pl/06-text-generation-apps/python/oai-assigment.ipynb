## Czym jest aplikacja do generowania tekstu?

Zwykle, gdy tworzysz aplikację, ma ona jakiś interfejs jak poniżej:

- Oparta na poleceniach. Aplikacje konsolowe to typowe aplikacje, w których wpisujesz polecenie, a ono wykonuje zadanie. Na przykład `git` to aplikacja oparta na poleceniach.
- Interfejs użytkownika (UI). Niektóre aplikacje mają graficzne interfejsy użytkownika (GUI), gdzie klikasz przyciski, wpisujesz tekst, wybierasz opcje i więcej.

### Aplikacje konsolowe i GUI są ograniczone

Porównaj to z aplikacją opartą na poleceniach, w której wpisujesz polecenie:

- **Jest ograniczona**. Nie możesz wpisać dowolnego polecenia, tylko te, które aplikacja obsługuje.
- **Specyficzna dla języka**. Niektóre aplikacje wspierają wiele języków, ale domyślnie aplikacja jest tworzona dla konkretnego języka, nawet jeśli możesz dodać wsparcie dla innych języków.

### Zalety aplikacji do generowania tekstu

Czym więc różni się aplikacja do generowania tekstu?

W aplikacji do generowania tekstu masz większą elastyczność, nie jesteś ograniczony do zestawu poleceń czy konkretnego języka wejściowego. Zamiast tego możesz używać języka naturalnego do interakcji z aplikacją. Kolejną zaletą jest to, że wchodzisz w interakcję z źródłem danych, które było trenowane na ogromnym korpusie informacji, podczas gdy tradycyjna aplikacja może być ograniczona do tego, co znajduje się w bazie danych.

### Co mogę zbudować za pomocą aplikacji do generowania tekstu?

Możesz zbudować wiele rzeczy. Na przykład:

- **Chatbot**. Chatbot odpowiadający na pytania o tematy, takie jak twoja firma i jej produkty, może być dobrym wyborem.
- **Pomocnik**. LLM są świetne w takich zadaniach jak podsumowywanie tekstu, uzyskiwanie wglądów z tekstu, generowanie tekstu takich jak CV i więcej.
- **Asystent kodu**. W zależności od używanego modelu językowego możesz zbudować asystenta kodu, który pomaga pisać kod. Na przykład możesz użyć takich produktów jak GitHub Copilot, jak również ChatGPT, by pomóc pisać kod.

## Generowanie tekstu

Sposobem generowania tekstu jest użycie metody `responses.create`. Oto przykład:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

W powyższym kodzie tworzymy obiekt odpowiedzi i przekazujemy model, którego chcemy użyć oraz prompt. Następnie drukujemy wygenerowany tekst.

### Odpowiedzi chatbota

Do tej pory widziałeś, jak generować tekst z pojedynczego promptu. API Responses jest również dobrze przystosowane do chatbotów - podajesz listę wiadomości jako `input`, aby zbudować rozmowę. Oto przykład użycia:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````


    - Tworzymy nowy prompt, dodając wynik z pierwszego promptu do nowego promptu:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Robimy nowe zapytanie, ale biorąc pod uwagę liczbę tokenów, o które prosiliśmy w pierwszym promptcie, więc tym razem ustawiamy `max_output_tokens` na 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Uruchamiając ten kod, otrzymujemy następujący wynik:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.
        
        -Apple and flour pancakes: 1 cup flour, 1/2 tsp baking powder, 1/2 tsp baking soda, 1/4 tsp salt, 1 tbsp sugar, 1 egg, 1 cup buttermilk or sour milk, 1/4 cup melted butter, 1 Granny Smith apple, peeled and grated
        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **Kilka słów o długości tokenów**. Powinniśmy zastanowić się, ile tokenów potrzebujemy, aby wygenerować tekst, który chcemy. Tokeny kosztują pieniądze, więc tam, gdzie to możliwe, powinniśmy starać się być oszczędni jeśli chodzi o ich liczbę. Na przykład, czy możemy sformułować prompt tak, aby użyć mniej tokenów?

   Aby zmienić użycie tokenów, możesz użyć parametru `max_output_tokens`. Na przykład, jeśli chcesz użyć 100 tokenów, zrobisz to tak:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Eksperymentowanie z temperaturą**. Temperatura to coś, o czym dotąd nie wspomnieliśmy, a jest to ważny kontekst wpływający na działanie programu. Im wyższa wartość temperatury, tym bardziej losowy będzie wynik. Natomiast im niższa wartość, tym bardziej przewidywalny będzie wynik. Zastanów się, czy chcesz mieć w swoim wyniku zróżnicowanie, czy nie.

   Aby zmienić temperaturę, możesz użyć parametru `temperature`. Na przykład, jeśli chcesz użyć temperatury 0.5, zrobisz to tak:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Uwaga, im bliżej 1.0, tym bardziej zróżnicowany będzie wynik.
