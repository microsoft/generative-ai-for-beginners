## Czym jest aplikacja do generowania tekstu?

Zazwyczaj, gdy tworzysz aplikację, ma ona jakiś interfejs, na przykład taki:

- Oparta na poleceniach. Aplikacje konsolowe to typowe aplikacje, w których wpisujesz polecenie i ono wykonuje zadanie. Na przykład `git` jest aplikacją opartą na poleceniach.
- Interfejs użytkownika (UI). Niektóre aplikacje mają graficzne interfejsy użytkownika (GUI), gdzie klikasz przyciski, wprowadzasz tekst, wybierasz opcje i więcej.

### Aplikacje konsolowe i UI mają ograniczenia

Porównaj to do aplikacji opartej na poleceniach, gdzie wpisujesz komendę: 

- **Są ograniczone**. Nie możesz wpisać dowolnej komendy, tylko te, które aplikacja obsługuje.
- **Specyficzne dla języka**. Niektóre aplikacje obsługują wiele języków, ale domyślnie aplikacja jest stworzona dla konkretnego języka, nawet jeśli możesz dodać wsparcie dla innych języków.

### Zalety aplikacji do generowania tekstu

Czym więc różni się aplikacja do generowania tekstu?

W aplikacji do generowania tekstu masz więcej elastyczności, nie jesteś ograniczony do zestawu poleceń czy konkretnego języka wejściowego. Możesz używać naturalnego języka do interakcji z aplikacją. Kolejną zaletą jest to, że wchodzisz w interakcję ze źródłem danych wytrenowanym na ogromnym korpusie informacji, podczas gdy tradycyjna aplikacja może być ograniczona do tego, co jest w bazie danych.

### Co mogę zbudować z aplikacją do generowania tekstu?

Możesz zbudować wiele rzeczy. Na przykład:

- **Chatbota**. Chatbot odpowiadający na pytania o tematy, takie jak twoja firma i jej produkty, może być dobrym rozwiązaniem.
- **Pomocnika**. LLM są świetne w takich zadaniach jak podsumowywanie tekstu, uzyskiwanie wglądów z tekstu, generowanie tekstu, np. CV i wiele więcej.
- **Asystenta kodu**. W zależności od używanego modelu językowego możesz zbudować asystenta kodu, który pomoże Ci pisać kod. Na przykład możesz użyć produktu takiego jak GitHub Copilot oraz ChatGPT, aby pomóc w pisaniu kodu.

## Generowanie tekstu

Sposobem na wygenerowanie tekstu jest użycie metody `responses.create`. Oto przykład:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

W powyższym kodzie tworzymy obiekt odpowiedzi, przekazujemy model, którego chcemy użyć oraz polecenie. Następnie drukujemy wygenerowany tekst.

### Odpowiedzi czatu

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

We now have a working app capable of producing recipes and it's flexible as it relies on inputs from the user, both on the number of recipes but also the ingredients used.

To further improve it, we want to add the following:

- **Filter out ingredients**. We want to be able to filter out ingredients we don't like or are allergic to. To accomplish this change, we can edit our existing prompt and add a filter condition to the end of it like so:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Above, we add `{filter}` to the end of the prompt and we also capture the filter value from the user.
