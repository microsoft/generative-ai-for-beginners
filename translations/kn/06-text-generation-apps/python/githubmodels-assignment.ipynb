# ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ನೀವು ಈ ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ಇದುವರೆಗೂ ಕಂಡಿದೆಯೇನಂದರೆ ಪ್ರಮುಖ ಸಂಯೋಜನೆಗಳು ಇವೆ, ಉದಾಹರಣೆಗೆ ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಎಂಜಿನಿಯರಿಂಗ್" ಎಂದು ಕರೆಯಲ್ಪಡುವ ಸಂಪೂರ್ಣ ವಿಭಾಗ. ನೀವು ಚಾಟ್‌ಜಿಪಿಟಿ, ಆಫೀಸ್ 365, ಮೈಕ್ರೋಸಾಫ್ಟ್ ಪವರ್ ಪ್ಲಾಟ್‌ಫಾರ್ಮ್ ಮತ್ತು ಇನ್ನಷ್ಟು ಉಪಕರಣಗಳೊಂದಿಗೆಸಂವಹನ ಮಾಡಬಹುದು, ಅವುಗಳಲ್ಲಿ ನೀವು ಏನನ್ನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸುವ ಮೂಲಕ ಬೆಂಬಲಿಸುತ್ತವೆ.

ನೀವು ಆ experiences್ನ್ನು ಅಪ್ಲಿಕೇಶನ್‌‌ಗೆ ಸೇರಿಸಲು, ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಗಳು ಮತ್ತು ಯಾವ್ ಗ್ರಂಥಾಲಯ ಬಳಸಬೇಕೆಂದು ಆಯ್ಕೆಮಾಡುವಂತಹ ಸಂಯೋಜನೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು. ಇವುಗಳನ್ನೆಲ್ಲಾ ನೀವು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ಕಲಿತುಕೊಳ್ಳುತ್ತೀರಿ.

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನೀವು:

- openai ಗ್ರಂಥಾಲಯ ಮತ್ತು ಅದರ ಪ್ರಮುಖ ಸಂಯೋಜನೆಗಳನ್ನು ತಿಳಿದುಕೊಳ್ಳಬೇಕು.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲು ಕಲಿತುಕೊಳ್ಳಬೇಕು.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟೋಕನ್‌ಗಳಂತಹ ಸಂಯೋಜನೆಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ನಿರ್ಮಿಸಬೇಕು ಎಂದು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು ಮಾಡಬಹುದಾದವುಗಳು:

- ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂದು ವಿವರಿಸಬಹುದು.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಬಹುದು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಟೋಕನ್‌ಗಳ ಬಳಕೆ ಹೆಚ್ಚಿಸುವುದಕ್ಕೆ ಅಥವಾ ಕಡಿಮೆ ಮಾಡುವುದಕ್ಕೆ ಮತ್ತು ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸಿ ವೈವಿಧ್ಯಮಯ ಫಲಿತಾಂಶಕ್ಕೆ ಹೊಂದಿಕೆಯಾಗಿಸಬಹುದು.

## ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುತ್ತಿದ್ದಾಗ ಅದರ ಕೆಲವೊಂದು ರೀತಿಯ ಇಂಟರ್ಫೇಸ್ ಇರುತ್ತದೆ, ಹೀಗಿದೆ:

- ಕಮಾಂಡ್ ಆಧಾರಿತ. ಕൺಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿದ್ದು, ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಿ ಆ ಕಾರ್ಯವನ್ನು ನಿರ್ವಹಿಸುತ್ತವೆ. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ ಗ್ರಾಫಿಕಲ್ ಯೂಸರ್ ಇಂಟರ್ಫೇಸ್ ಗಳು (GUI) ಇರುತ್ತವೆ, ಅವುಗಳಲ್ಲಿ ನೀವು ಬಟನ್‌ಗಳನ್ನು ಕ್ಲಿಕ್ ಮಾಡಬಹುದು, ಪಠ್ಯ ನುಡಿ ತುಂಬಬಹುದು, ಆಯ್ಕೆಗಳನ್ನು ಆರಿಸಬಹುದು ಮತ್ತು ಇನ್ನಷ್ಟು.

### ಕಂಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ ಸೀಮಿತತೆಗಳಿವೆ

ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್ ನಡಿವಾಟನ್ನು ಹೋಲಿಕೆ ಮಾಡಿ, ನೀವು ಕಮಾಂಡ್ ಅನ್ನು ಟೈಪ್ ಮಾಡುತ್ತೀರಿ:

- **ಇದಕ್ಕೆ ಸೀಮಿತತೆ ಇದೆ**. ನೀವು ಯಾವುದೇ ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಲು ಸಾಧ್ಯವಿಲ್ಲ, ಕೇವಲ ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಕಮಾಂಡ್‌ಗಳಿಗಷ್ಟೆ ಸೀಮಿತವಾಗಿದೆ.
- **ಭಾಷಾ ನಿರ್ದಿಷ್ಟ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಅನೇಕ ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಬೇಸಿಗೆ ಆ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗಾಗಿ ನಿರ್ಮಿಸಲಾಗಿದೆ, ನೀವು ಹೆಚ್ಚು ಭಾಷಾ ಬೆಂಬಲವನ್ನು ಸೇರಿಸಬಹುದು ಎಂದರೂ ಸಹ.

### ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಪ್ರಯೋಜನಗಳು

ಹೀಗಾಗಿ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಹೇಗೆ ವಿಭಿನ್ನವಾಗಿವೆ?

ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ, ನೀವು ಇನ್ನಷ್ಟು ಸುಲಭತೆಯೊಂದಿಗೆ ಇರಬೇಕು, ನೀವು ಕಮಾಂಡ್‌ಗಳು ಅಥವಾ ನಿರ್ದಿಷ್ಟ ಇನ್ಪುಟ್ ಭಾಷೆಯ ಸೀಮಿತತೆಗೆ ಸೀಮಿತವಾಗಿದ್ದೀರಿ. ಬದಲಾಗಿ, ನೀವು ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಅಪ್ಲಿಕೇಶನ್‌ ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಬಹುದು. ಮತ್ತೊಂದು ಪ್ರಯೋಜನವೆಂದರೆ, ನೀವು ಈಗಾಗಲೇ ಒಂದು ಬೃಹತ್ ಮಾಹಿತಿಯ ಸಂಗ್ರಹವನ್ನು ತರಬೇತುದಾನೆಂದ ಹಾಗಿರುವ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತಿದ್ದೀರಿ, ಹೀಗಾಗಿ ಸಾಂಪ್ರದಾಯಿಕ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿನ ವಿಷಯಗಳಿಗೆ ಸೀಮಿತವಾಗಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ನೊಂದಿಗೆ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಅನೇಕ ವಸ್ತುಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ఉదಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಬಗ್ಗೆ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಹೊಂದಾಣಿಕೆಯಾಗಬಹುದು.
- **ಸಹಾಯಕ**. LLMಗಳು ಪಠ್ಯ ಸಾರಾಂಶ ಮಾಡುವುದು, ಪಠ್ಯದಿಂದ洞察ಗಳನ್ನು ಪಡೆಯುವುದು, ಜೀವವಿವರಗಳಂತಹ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವುದು ಮುಂತಾದ ಕೆಲಸಗಳಲ್ಲಿ ತೊಡಗಿಸಿಕೊಂಡಿವೆ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬിച്ചೆಯೇ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕನನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಅನ್ನು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಆಗಿ ಬಳಸಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ನಿವು LLM ಮೂಲಕ ಸಂಯೋಜಿಸುವ ಮಾರ್ಗವನ್ನು ಕಂಡುಹಿಡಿಯಬೇಕು, ಸಾಮಾನ್ಯವಾಗಿ ಕೆಳಗಿನ ಎರಡು ವಿಧಾನಗಳು ಇವೆ:

- API ಬಳಸುವುದು. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ನೊಂದಿಗೆ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಗ್ರಂಥಾಲಯವನ್ನು ಬಳಸುವುದು. ಗ್ರಂಥಾಲಯಗಳು API ಕರೆಗಳನ್ನು ಒಳಗೊಳಿಸುವ ಮೂಲಕ ಉಪಯೋಗಿಸಲು ಸುಲಭವಾಗಿಸುತ್ತವೆ.

## ಗ್ರಂಥಾಲಯಗಳು/SDKಗಳು

LLMಗಳನ್ನು ಬಳಸಿ ಕೆಲಸ ಮಾಡುವ ಕೆಲವು ಪ್ರಸಿದ್ಧ ಗ್ರಂಥಾಲಯಗಳು ಇವೆ:

## ChatCompletions ಮೂಲಕ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸಿ

ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವ ವಿಧಾನವೆಂದರೆ `ChatCompletionsClient` ವರ್ಗವನ್ನು ಬಳಸುವುದು.
`samples/python/azure_ai_inference/basic.py` ನಲ್ಲಿ, ಪ್ರತಿಕ್ರಿಯೆ ವಿಭಾಗದಲ್ಲಿನ ಕೋಡ್‌ನಲ್ಲಿ ಬಳಕೆದಾರರ ಪಾತ್ರವನ್ನು ತಿದ್ದುಪಡಿ ಮಾಡಿ, ಕೆಳಗಿನಂತೆ content ಪರಿಮಾಣವನ್ನು ಬದಲಾಯಿಸಿ:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ಐಚ್ಛಿಕ ಪ್ಯಾರಾಮೀಟರ್ಗಳು
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

ಹೊಸ ಹುರಿಗೆಯನ್ನು ಚಾಲನೆ ಮಾಡಿ ಫಲಿತಾಂಶವನ್ನು ನೋಡಿ


## ವಿಭಿನ್ನ ಕಾರ್ಯಗಳಿಗೆ ವಿಭಿನ್ನ ಬಗೆಯ ಪ್ರಾಂಪ್ಟ್‌ಗಳು

ನೀವು ಈಗಾಗಲೆಲ್ಲ ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಪಠ್ಯವನ್ನು ಹೇಗೆ ಹಂಪಿಸಲು ಮತ್ತು ಉತ್ಪಾದಿಸಲು ಗೊತ್ತಾಗಿದೆ. ನೀವು ವಿಭಿನ್ನ ಬಗೆಯ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸಲು ಬದಲಾಯಿಸಬಹುದಾದ ಹಾಗೂ ರಚಿಸಬಹುದಾದ ಕಾರ್ಯಕ್ರಮವನ್ನೂ ಹೊಂದಿದ್ದೀರಿ.

ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ವಿವಿಧ ಕೆಲಸಗಳಿಗೆ ಉಪಯೋಗಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಒಂದು ಬಗೆಯ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದನೆ ಮಾಡುವುದು**. ಉದಾಹರಣೆಗೆ, ನೀವು ಪೋಎಮ್, ಕ್ವಿಜ್ ಪ್ರಶ್ನೆಗಳು ಮುಂತಾದವುಗಳನ್ನು ರಚಿಸಬಹುದು.
- **ಮಾಹಿತಿಯನ್ನು ಹುಡುಕುವುದು**. ನೀವು 'ವೆಬ್ ಅಭಿವೃದ್ಧಿಯಲ್ಲಿ CORS ಅರ್ಥವೇನು?' ಎಂಬುದಾಗಿ ಹೋಲುವಂತೆ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ಮೂಲಕ ಮಾಹಿತಿ ಹುಡುಕಬಹುದು.
- **ಕೋಡ್ ಉತ್ಪಾದನೆ ಮಾಡುವುದು**. ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳ ಮೂಲಕ ಮೀಲ್‌ಗಳನ್ನು ಪರಿಶೀಲಿಸಲು ಉಪಯೋಗಿಸುವ ನಿಯಮಿತ ಅಭಿವ್ಯಕ್ತಿ ಅಥವಾ ಪೂರ್ಣ ಕಾರ್ಯಕ್ರಮ, ಜಾಲ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಕೌುರ್ ಮಾಡಬಹುದು.

## ವ್ಯಾಯಾಮ - ರೆಸಿಪಿ ಜನರೆಟರ್ ನಿರ್ಮಿಸಿ

ನಾವು ಈಗಾಗಲೇ ಒಂದು ಘಟನಾಚರಣೆಯನ್ನು ಅನುಭವಿಸಿದ್ದೇವೆ, ಅದಕ್ಕೆ ಹೊಂದಿಕೊಂಡಂತೆ ಕೋಡ್ ಅನ್ನು ಬರೆಯೋಣ. ಇದಕ್ಕಾಗಿ ಈ ಹಂತಗಳನ್ನು ಅನುಸರಿಸಿ:

1. ಇತ್ತೀಚಿನ ಫೈಲ್ ಅನ್ನು ಆರಂಭಿಕ ಬಿಂದುವಾಗಿ ಬಳಸಿ
1. ಒಂದು `prompt` ಚರವನ್ನು ರಚಿಸಿ ಮತ್ತು ಕೆಳಗಿನಂತೆ ಮಾದರಿ ಕೋಡ್ ಅನ್ನು ಬದಲಾಯಿಸಿ:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
