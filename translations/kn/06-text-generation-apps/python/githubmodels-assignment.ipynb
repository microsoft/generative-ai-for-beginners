# ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ನೀವು ಈ ಪಠ್ಯಕ್ರಮದ ಮೂಲಕ ಈಗಾಗಲೇ ನೋಡಿದ್ದೀರಿ, ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಎಂಜಿನಿಯರಿಂಗ್" ಎಂಬ ಸಂಪೂರ್ಣ ಶಿಸ್ತಿನಂತಹ ಮೂಲಭೂತ ಕಲ್ಪನೆಗಳಿವೆ. ನೀವು ಸಂವಹನ ಮಾಡಬಹುದಾದ ಅನೇಕ ಸಾಧನಗಳು, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮತ್ತು ಇನ್ನಷ್ಟು, ನೀವು ಏನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಲು ಬೆಂಬಲಿಸುತ್ತವೆ.

ನೀವು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಇಂತಹ ಅನುಭವವನ್ನು ಸೇರಿಸಲು, ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಗಳು ಮತ್ತು ಕೆಲಸ ಮಾಡಲು ಲೈಬ್ರರಿ ಆಯ್ಕೆಮಾಡುವುದು ಎಂಬ ಕಲ್ಪನೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕಾಗುತ್ತದೆ. ಇದೇನೂ ನೀವು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ಕಲಿಯಲಿದ್ದೀರಿ.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನೀವು:

- openai ಲೈಬ್ರರಿ ಮತ್ತು ಅದರ ಮೂಲಭೂತ ಕಲ್ಪನೆಗಳ ಬಗ್ಗೆ ತಿಳಿಯಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟೋಕನ್‌ಗಳಂತಹ ಕಲ್ಪನೆಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವ ವಿಧಾನವನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಿ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು:

- ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂದು ವಿವರಿಸಬಹುದು.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಬಹುದು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ಹೆಚ್ಚು ಅಥವಾ ಕಡಿಮೆ ಟೋಕನ್‌ಗಳನ್ನು ಬಳಸಲು ಮತ್ತು ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸಲು ಸಂರಚಿಸಬಹುದು, ವಿಭಿನ್ನ ಔಟ್‌ಪುಟ್‌ಗಾಗಿ.

## ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವಾಗ, ಅದು ಕೆಳಗಿನಂತೆಯಾದ ಇಂಟರ್ಫೇಸ್ ಹೊಂದಿರುತ್ತದೆ:

- ಕಮಾಂಡ್ ಆಧಾರಿತ. ಕಾನ್ಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಿ ಕಾರ್ಯವನ್ನು ನಿರ್ವಹಿಸುವ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (GUI) ಹೊಂದಿವೆ, ಅಲ್ಲಿ ನೀವು ಬಟನ್‌ಗಳನ್ನು ಕ್ಲಿಕ್ ಮಾಡಬಹುದು, ಪಠ್ಯವನ್ನು ನಮೂದಿಸಬಹುದು, ಆಯ್ಕೆಯನ್ನು ಆರಿಸಬಹುದು ಮತ್ತು ಇನ್ನಷ್ಟು.

### ಕಾನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸೀಮಿತ

ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡುವ ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್‌ಗಳೊಂದಿಗೆ ಹೋಲಿಸಿ ನೋಡಿ:

- **ಇದು ಸೀಮಿತವಾಗಿದೆ**. ನೀವು ಯಾವುದೇ ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಲಾಗುವುದಿಲ್ಲ, ಕೇವಲ ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಕಮಾಂಡ್‌ಗಳು ಮಾತ್ರ.
- **ಭಾಷಾ ನಿರ್ದಿಷ್ಟ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಅನೇಕ ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಡೀಫಾಲ್ಟ್‌ನಲ್ಲಿ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗೆ ನಿರ್ಮಿಸಲಾಗಿದೆ, ನೀವು ಹೆಚ್ಚಿನ ಭಾಷಾ ಬೆಂಬಲವನ್ನು ಸೇರಿಸಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಲಾಭಗಳು

ಹೀಗಾಗಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ವಿಭಿನ್ನ?

ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ, ನಿಮಗೆ ಹೆಚ್ಚು ಸ್ವಾತಂತ್ರ್ಯವಿದೆ, ನೀವು ಕಮಾಂಡ್‌ಗಳ ಒಂದು ಸೆಟ್ ಅಥವಾ ನಿರ್ದಿಷ್ಟ ಇನ್‌ಪುಟ್ ಭಾಷೆಗೆ ಸೀಮಿತವಾಗಿಲ್ಲ. ಬದಲಾಗಿ, ನೀವು ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಅಪ್ಲಿಕೇಶನ್ ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಬಹುದು. ಮತ್ತೊಂದು ಲಾಭವೆಂದರೆ, ನೀವು ಈಗಾಗಲೇ ವಿಶಾಲ ಮಾಹಿತಿಯ ಸಂಗ್ರಹದ ಮೇಲೆ ತರಬೇತುಗೊಂಡಿರುವ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತಿದ್ದೀರಿ, ಆದರೆ ಸಾಂಪ್ರದಾಯಿಕ ಅಪ್ಲಿಕೇಶನ್ ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿರುವ ಮಾಹಿತಿಗೆ ಸೀಮಿತವಾಗಿರಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ನೊಂದಿಗೆ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಅನೇಕ ವಸ್ತುಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಬಗ್ಗೆ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಆಯ್ಕೆಯಾಗಬಹುದು.
- **ಸಹಾಯಕ**. LLMಗಳು ಪಠ್ಯ ಸಾರಾಂಶ, ಪಠ್ಯದಿಂದ ಒಳನೋಟಗಳನ್ನು ಪಡೆಯುವುದು, ರೆಸ್ಯೂಮ್‌ಗಳಂತಹ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವುದು ಮತ್ತು ಇನ್ನಷ್ಟು ಕಾರ್ಯಗಳಲ್ಲಿ ಉತ್ತಮವಾಗಿವೆ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿಸಿ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಅನ್ನು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡಲು ಬಳಸಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ನೀವು ಸಾಮಾನ್ಯವಾಗಿ ಕೆಳಗಿನ ಎರಡು ವಿಧಾನಗಳನ್ನು ಒಳಗೊಂಡಿರುವ LLM ಜೊತೆಗೆ ಸಂಯೋಜಿಸುವ ಮಾರ್ಗವನ್ನು ಹುಡುಕಬೇಕಾಗುತ್ತದೆ:

- API ಬಳಸಿ. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ನೊಂದಿಗೆ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ, ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಲೈಬ್ರರಿ ಬಳಸಿ. ಲೈಬ್ರರಿಗಳು API ಕರೆಗಳನ್ನು ಸಂಗ್ರಹಿಸಿ ಅವುಗಳನ್ನು ಬಳಸಲು ಸುಲಭವಾಗಿಸುತ್ತವೆ.

## ಲೈಬ್ರರಿಗಳು/SDKಗಳು

## ವಿಭಿನ್ನ ಕಾರ್ಯಗಳಿಗೆ ವಿಭಿನ್ನ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ಪ್ರಕಾರಗಳು

ನೀವು ಈಗ ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಪಠ್ಯವನ್ನು ಹೇಗೆ ರಚಿಸುವುದನ್ನು ನೋಡಿದ್ದೀರಿ. ನೀವು ಬದಲಾಯಿಸಬಹುದಾದ ಮತ್ತು ವಿಭಿನ್ನ ರೀತಿಯ ಪಠ್ಯವನ್ನು ರಚಿಸಲು ಬಳಸಬಹುದಾದ ಒಂದು ಪ್ರೋಗ್ರಾಂ ಕೂಡ ಹೊಂದಿದ್ದೀರಿ.

ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಎಲ್ಲಾ ರೀತಿಯ ಕಾರ್ಯಗಳಿಗೆ ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಒಂದು ರೀತಿಯ ಪಠ್ಯವನ್ನು ರಚಿಸಿ**. ಉದಾಹರಣೆಗೆ, ನೀವು ಒಂದು ಕವಿತೆ, ಪ್ರಶ್ನೋತ್ತರಗಳಿಗಾಗಿ ಪ್ರಶ್ನೆಗಳು ಇತ್ಯಾದಿ ರಚಿಸಬಹುದು.
- **ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಿ**. ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಕೆಳಗಿನ ಉದಾಹರಣೆಯಂತೆ 'ವೆಬ್ ಅಭಿವೃದ್ಧಿಯಲ್ಲಿ CORS ಎಂದರೇನು?' ಎಂಬ ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಬಹುದು.
- **ಕೋಡ್ ರಚಿಸಿ**. ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಕೋಡ್ ರಚಿಸಬಹುದು, ಉದಾಹರಣೆಗೆ ಇಮೇಲ್‌ಗಳನ್ನು ಪರಿಶೀಲಿಸಲು ಬಳಸುವ ನಿಯಮಿತ ಅಭಿವ್ಯಕ್ತಿ ರಚಿಸುವುದು ಅಥವಾ ಸಂಪೂರ್ಣ ಪ್ರೋಗ್ರಾಂ, ಉದಾಹರಣೆಗೆ ವೆಬ್ ಅಪ್ಲಿಕೇಶನ್ ರಚಿಸುವುದು.

## ವ್ಯಾಯಾಮ - ಒಂದು ರೆಸಿಪಿ ಜನರೇಟರ್ ನಿರ್ಮಿಸಿ

ಈಗ ನಾವು ಒಂದು ದೃಶ್ಯಾವಳಿಯನ್ನು ಆಡಿದಿದ್ದೇವೆ, ಪ್ರದರ್ಶಿತ ದೃಶ್ಯಾವಳಿಗೆ ಹೊಂದಿಕೊಳ್ಳುವ ಕೋಡ್ ಅನ್ನು ಬರೆಯೋಣ. ಅದಕ್ಕಾಗಿ, ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸಿ:

1. ಇತ್ತೀಚಿನ ಫೈಲ್ ಅನ್ನು ಪ್ರಾರಂಭಿಕ ಬಿಂದುವಾಗಿ ಬಳಸಿ
1. `prompt` ಚರವನ್ನು ರಚಿಸಿ ಮತ್ತು ಕೆಳಗಿನಂತೆ ಮಾದರಿ ಕೋಡ್ ಅನ್ನು ಬದಲಾಯಿಸಿ:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
