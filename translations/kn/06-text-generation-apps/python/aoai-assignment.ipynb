# ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ಈ ಪಠ್ಯಕ್ರಮದ ಮೂಲಕ ನೀವು ಈಗಾಗಲೇ ನೋಡಿದ್ದೀರಿ ಅಲ್ಲಿ ಪ್ರಮುಖ ಕಲ್ಪನೆಗಳು ಇದ್ದಿವೆ - ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಎಂಜಿನಿಯರಿಂಗ್" ಎಂದು ಕರೆಯಲ್ಪಡುವ ಸಂಪೂರ್ಣ ವಸತುಶಾಸ್ತ್ರವೂ ಇದೆ. ನೀವು ಸಂವಾದಿಸಬಹುದಾದ ಹಲವಾರು ಸಲಕರಣೆಗಳು ಇದ್ದಾರೆ, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮುಂತಾದವುಗಳು, ನೀವು ಏನನ್ನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಲು ನಿಮ್ಮನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ.

ಮುಖ್ಯವು ಎಂದರೆ, ನೀವು ಆಂತಹ ಅನುಭವವನ್ನು ಏಪ್ನಲ್ಲಿ ಸೇರಿಸಲು, ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ ಮತ್ತು ಕೆಲಸ ಮಾಡಲು ಲೈಬ್ರರಿ ಆಯ್ಕೆ ಮಾಡಿಕೊಳ್ಳುವುದನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು. ಇದು ನೀವು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ಕಲಿಯುವದು.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ನೀವು:

- openai ಲೈಬ್ರರಿ ಮತ್ತು ಅದರ ಮುಖ್ಯ ಸಂಪ್ರದಾಯಗಳ ಬಗ್ಗೆ ಕಲಿಯಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ ನಿರ್ಮಿಸಿ.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟೋಕನ್ಗಳಂತಹ ಕಲ್ಪನೆಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ ಅನ್ನು ನಿರ್ಮಿಸುವ ವಿಧಾನವನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಿ.

## ಕಲಿಕೆ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು ಶಕ್ತರಾಗಿರುತ್ತೀರಿ:

- ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂಬುದನ್ನು ವಿವರಿಸಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿರಿ.
- ಹಲವು ಅಥವಾ ಕಡಿಮೆ ಟೋಕನ್ಗಳನ್ನು ಬಳಸಲು ಮತ್ತು ವಿವಿಧ ಔಟ್‌ಪುಟ್‌ಗಾಗಿ ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸಲು ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಕಾನ್ಫಿಗರ್ ಮಾಡಿರಿ.

## ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವಾಗ ಇದಕ್ಕೆ ಕೆಳಗಿನಂತೆಯೇ ಯಾವುದೋ ಒಬ್ಬ ಇಂಟರ್ಫೇಸ್ ಇರುತ್ತದೆ:

- ಕಮಾಂಡ್ ಆಧಾರಿತ. ಕನ್ಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿದ್ದು ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡುತ್ತೀರಿ ಹಾಗು ಅದು ಕೆಲಸವನ್ನು ನಿರ್ವಹಿಸುತ್ತದೆ. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (UI). ಕೆಲ ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (GUI) ಗಳಿದ್ದು ನೀವು ಬಟನ್ಗಳನ್ನು ಕ್ಲಿಕ್ ಮಾಡಬಹುದು, ಪಠ್ಯವನ್ನು ನಮೂದಿಸಬಹುದು, ಆಯ್ಕೆಗಳನ್ನು ಆರಿಸಬಹುದು ಮುಂತಾದವು.

### ಕನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳು పరಿಮಿತ

ನೀವು ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್ ಜೊತೆಗೆ ಹೋಲಿಸಿ: 

- **ಇದು పరಿಮಿತವಾಗಿದೆ**. ನೀವು ಯಾವದ್ದು ಬೇಕಾದರೂ ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಲಾಗದು, ಕೇವಲ ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಕಮಾಂಡ್‌ಗಳನ್ನು ಮಾತ್ರ.
- **ಭಾಷಾ ನಿರ್ದಿಷ್ಟ**. ಕೆಲ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಹಲವಾರು ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಮೂಲತಃ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗೆ ನಿರ್ಮಿಸಲಾಗಿದೆ, ಹೆಚ್ಚುವರಿ ಭಾಷೆ ಬೆಂಬಲವು ಸೇರಿಸಬಹುದಾದರೂ.

### ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಲಾಭಗಳು

ಹಾಗಿದ್ರೆ ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಬೇರೆ ಹೇಗೆ?

ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ, ನಿಮಗೆ ಬಹುಮಟ್ಟಿನ ಸಡಸರ ವಿನ್ಯಾಸವಿದೆ, ನೀವು ಕಮಾಂಡ್‌ಗಳ ಅಥವಾ ನಿರ್ದಿಷ್ಟ ಭಾಷೆಯ ಒಳಗೆ ಸೀಮಿತರಾಗಿರುವುದಿಲ್ಲ. ಬದಲಿಗೆ, ನೀವು ನೈಸರ್ಗಿಕ ಭಾಷೆ ಬಳಸಿ ಅಪ್ಲಿಕೇಶನ್ ಸಹಿತ ಸಂವಹನ ಮಾಡಬಹುದು. ಇನ್ನೊಂದು ಲಾಭವೆಂದರೆ ನೀವು ಈಗಾಗಲೇ ಮಹತ್ತರ ಮಾಹಿತಿಯ ಸಂಗ್ರಹದ ಮೇಲೆ ತರಬೇತಿ ಪಡೆದಿರುವ ಡೇಟಾ ಮೂಲ ಸೋಮಾಮಾತ್ರೆಯನ್ನು ಸಂವಹನ ಮಾಡುತ್ತೀರಿ, ಆದರೆ ಪರಂಪರাগত ಅಪ್ಲಿಕೇಶನ್ ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿ ಇರುವುದಕ್ಕೆ ಸೀಮಿತವಾಗಿರಬಹುದು. 

### ಪಠ್ಯ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಜೊತೆ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಬಹಳ ಕೆಲವು ವಿಷಯಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ಸಂಸ್ಥೆ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಬಗ್ಗೆ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಹೊಂದಾಣಿಕೆ ಆಗಬಹುದು.
- **ಸಹಾಯಕ**. LLMಗಳು ಪಠ್ಯದ ಸಾರಾಂಶ ಮಾಡುವುದು, ಪಠ್ಯದಿಂದ ಒಳನೋಟ ಪಡೆದುವುದು, ರೆಸ್ಯೂಮ್ ಮುಂತಾದವುಗಳನ್ನು ರಚಿಸುವುದರಲ್ಲಿ ಉತ್ತಮವಾಗಿವೆ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿಸಿ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಮುಂತಾದ ಉತ್ಪನ್ನಗಳನ್ನು ಬಳಸಿ ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ಬಹುಶಃ, ನೀವು ಸಾಮಾನ್ಯವಾಗಿ ಕೆಳಗಿನ ಎರಡು ರೀತಿಗಳಲ್ಲಿ LLM ಜೊತೆಗೆ ಸಮನ್ವಯ ಮಾಡಿಕೊಳ್ಳಬೇಕಾಗುತ್ತದೆ:

- API ಉಪಯೋಗಿಸಿ. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಜೊತೆಗೆ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸುತ್ತೀರಿ ಮತ್ತು ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಲೈಬ್ರರಿ ಉಪಯೋಗಿಸಿ. ಲೈಬ್ರರಿಗಳು API ಕರೆಗಳನ್ನು ಮುಚ್ಚಿ ಸುಲಭವಾಗಿ ಬಳಸಲು ಸಹಾಯ ಮಾಡುತ್ತವೆ.

## ಲೈಬ್ರರಿಗಳು/SDKಗಳು

LLM ಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು ಕೆಲ ಪ್ರಸಿದ್ಧ ಲೈಬ್ರರಿಗಳು ಇವೆ:

- **openai**, ಈ ಲೈಬ್ರರಿ ನಿಮ್ಮ ಮಾದರಿಯನ್ನು ಸಂಪರ್ಕಿಸತಕ್ಕಂತೆ ಮತ್ತು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಕಳುಹಿಸುವುದನ್ನು ಸುಲಭ ಮಾಡುತ್ತದೆ.

ನಂತರ, ಉನ್ನತ ಮಟ್ಟದಲ್ಲಿ ಕಾರ್ಯನಿರ್ವಹಿಸುವ ಕೆಲವು ಲೈಬ್ರರಿಗಳಿವೆ:

- **Langchain**. Langchain ಹೆಚ್ಚು ಪ್ರಖ್ಯಾತವಾಗಿದೆ ಮತ್ತು ಪೈಥಾನ್ ಅನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Semantic Kernel**. Semantic Kernel ಮೈಕ್ರೋಸಾಫ್ಟ್ ನೀಡಿರುವ ಲೈಬ್ರರಿ, ಇದು C#, Python ಮತ್ತು Java ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.

## openai ಬಳಸಿ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್

ನಾವು ನಮ್ಮ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ನಿರ್ಮಿಸಬಹುದು, ಯಾವ ಲೈಬ್ರರಿಗಳು ಬೇಕು, ಎಷ್ಟು ಅಗತ್ಯವಿದೆ ಮತ್ತು ಇತ್ಯಾದಿಗಳನ್ನು ನೋಡೋಣ.

### openai ಸ್ಥಾಪನೆ

  > [!NOTE] Codespaces ಅಥವಾ Devcontainer ನಲ್ಲಿ ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು ಚಲಿಸಲು ಈ ಹಂತ ಅಗತ್ಯವಿಲ್ಲ


OpenAI ಅಥವಾ Azure OpenAI ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಲು ಹಲವಾರು ಲೈಬ್ರರಿಗಳು ಇವೆ. C#, Python, JavaScript, Java ಮತ್ತು ಇನ್ನಿತರ ಅನೇಕ ಪ್ರೋಗ್ರಾಮಿಂಗ್ ಭಾಷೆಗಳಲ್ಲಿಯೂ ಇದನ್ನು ಬಳಸಬಹುದು.  
ನಾವು `openai` ಪೈಥಾನ್ ಲೈಬ್ರರನ್ನು ಬಳಸುವ ನಿರ್ಧಾರ ತೆಗೆದುಕೊಂಡಿದ್ದೇವೆ, ಆದ್ದರಿಂದ ಅದನ್ನು ಸ್ಥಾಪಿಸಲು `pip` ಉಪಯೋಗಿಸುವೆವು.

```bash
pip install openai
```

ನೀವು Codespaces ಅಥವಾ Dev Container ನಲ್ಲಿ ಈ ನೋಡ್ ಬುಕನ್ನು ಚಲಿಸುತ್ತಿಲ್ಲವಾದರೆ, ನಿಮ್ಮ ಯಂತ್ರದಲ್ಲಿ [Python](https://www.python.org/) ಸ್ಥಾಪಿಸಿಕೊಳ್ಳಬೇಕಾಗುತ್ತದೆ.

### ಸಂಪನ್ಮೂಲ ರಚನೆ

ನೀವು ಈಗಾಗಲೇ ಮಾಡಿರಲಿಲ್ಲದಿದ್ದರೆ, ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸಬೇಕು:

- <https://azure.microsoft.com/free/> ನಲ್ಲಿ ಅಕೌಂಟ್ ತೆರೆಯಿರಿ.
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) ನಲ್ಲಿ Azure OpenAI ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸಿ. ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸುವ ವಿಧಾನವನ್ನು ನೋಡಲು ಈ ಮಾರ್ಗದರ್ಶಿಯನ್ನು ಅನುಸರಿಸಿ [create a resource](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst).


### API ಕೀ ಮತ್ತು ಎಂಡ್ಪಾಯಿಂಟ್‌ ಅನ್ನು ಕಂಡುಹಿಡಿಯಿರಿ

ಈ ಸಮಯದಲ್ಲಿ, ನೀವು ನಿಮ್ಮ `openai` ಲೈಬ್ರರಿಗೆ ಯಾವ API ಕೀ ಉಪಯೋಗಿಸುವುದೆಂದು ತಿಳಿಸಬೇಕು. ನಿಮ್ಮ API ಕೀ ಕಂಡುಹಿಡಿಯಲು, Azure OpenAI ಸಂಪನ್ಮೂಲದ "Keys and Endpoint" ವಿಭಾಗಕ್ಕೆ ಹೋದ ಮೇಲೆ "Key 1" ಮೌಲ್ಯವನ್ನು ಪ್ರತಿಯಲ್ಲಿ ಮಾಡಿ.

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

### Azure ಸಂರಚನೆಯ ಸೆಟ್‌ಅಪ್

ನೀವು Azure OpenAI ಬಳಸುತ್ತಿದ್ದರೆ, ಸಂರಚನೆಯನ್ನು ಈ ರೀತಿಯಲ್ಲಿ ಸೆಟ್‌ಅಪ್ ಮಾಡಬಹುದು. Responses API ಅನ್ನು Azure OpenAI **v1 ಎಂಡ್ಪಾಯಿಂಟ್** ನಿಂದ ಸೇವೆ ಮಾಡಲಾಗುತ್ತದೆ, ಆದ್ದರಿಂದ ನಾವು `OpenAI` ಕ್ಲೈಂಟ್ ಅನ್ನು `<your-endpoint>/openai/v1/` ಗೆ ಸೂಚಿಸುತ್ತೇವೆ:

```python
client = OpenAI(
```

ಮೇಲಿನ ಹಂತದಲ್ಲಿ ನಾವು ಕೆಳಕಂಡವುಗಳನ್ನು ಸೆಟ್ ಮಾಡುತ್ತಿದ್ದೇವೆ:

- `api_key`, ಇದು ನಿಮ್ಮ API ಕೀ, ಇದು Azure ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ಇದೆ.
- `base_url`, ಇದು ನಿಮ್ಮ Azure OpenAI ಎಂಡ್ಪಾಯಿಂಟ್ ಇದಕ್ಕೊಡನೆ `/openai/v1/` ಸೇರಿಸಲಾಗಿದೆ. ನೀವು API ಕೀಲಿಗೆ ಪಕ್ಕದಲ್ಲಿರುವ ಎಂಡ್ಪಾಯಿಂಟ್ ಅನ್ನು Azure ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ನೋಡಬಹುದು. v1 ಎಂಡ್ಪಾಯಿಂಟ್ ಬಳಸುವುದರಿಂದ ನಿಮಗೆ `api_version` ಅನ್ನು ಹಂಚಬೇಕಾಗುವುದಿಲ್ಲ.
- `deployment`, ಇದು ನೀವು Foundry ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ರಚಿಸಿದ ಮಾದರಿ ನಿಯೋಜನೆಯ ಹೆಸರು.

## ವಿಭಿನ್ನ ಕಾರ್ಯಗಳಿಗೆ ವಿಭಿನ್ನ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ವಿಧಗಳು

ಈಗ ನೀವು ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಪಠ್ಯವನ್ನು ಹೇಗೆ ಉತ್ಪಾದಿಸುವುದು ಎಂದು ನೋಡಿ. ನೀವು ಬದಲಾವಣೆಮಾಡಿ ವಿಭಿನ್ನ ವಿಧದ ಪಠ್ಯಗಳನ್ನು ನಿರ್ಮಿಸಲು ಒಂದು ಕಾರ್ಯಕ್ರಮವನ್ನು ಹೊಂದಿದ್ದೀರಾ.

ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಎಲ್ಲ ಬಗೆಯ ಕಾರ್ಯಗಳಿಗೆ ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಒಂದು ವಿಧದ ಪಠ್ಯವನ್ನು ರಚಿಸಿ**. ಉದಾಹರಣೆಗೆ, ನೀವು ಒಂದು ಕಾವ್ಯ, ಪ್ರಶ್ನೋತ್ತರವೇ್ಲು ಅಥವಾ ಇತರವುಗಳನ್ನು ರಚಿಸಬಹುದು.
- **ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಿ**. ಇಂತಹ ಉದಾಹರಣೆಯಲ್ಲಿರುವಂತೆ ವಿನಂತಿಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಮಾಹಿತಿ ಹುಡುಕಬಹುದು: 'ವೆಬ್ ಅಭಿವೃದ್ಧಿಯಲ್ಲಿ CORS ಎಂದರೇನು?'.
- **ಕೋಡ್ ರಚಿಸಿ**. ಇಮೇಲ್ ಪರಿಶೀಲನೆಗಾಗಿ ನಿಯಮಿತ ಭಿನ್ನಮುದ್ರಣ ತಯಾರಿಸುವಂತೆ ಕೋಡ್ ತಂದೆ ಮಾಡುವ ಕೋಡ್ ರಚನೆಗೆ ಪ್ರಾಂಪ்ட்‌ಗಳನ್ನು ಬಳಸಬಹುದು ಅಥವಾ ಸಂಪೂರ್ಣ ಕಾರ್ಯಕ್ರಮ ಬರೆದರೂ ಚೆನ್ನಾಗಿರುತ್ತದೆ, ಉದಾಹರಣೆಗೆ ವೆಬ್ ಅಪ್ಲಿಕೇಶನ್.
