# ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ನೀವು ಇದುವರೆಗೆ ಈ ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ನೋಡಿದಂತೆ, ಪ್ರಾಂಪ್ಟ್‌ಗಳಂತಹ ಹಿರಿತ ಸಂಪ್ರದಾಯಗಳಿವೆ ಹಾಗೂ "prompt engineering" ಎಂಬ ಸಂಪೂರ್ಣ ಶಿಸ್ತೂ ಇದೆ. ನೀವು ಸಂವಹನ ಮಾಡುವ ಹಲವಾರು ಉಪಕರಣಗಳು, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮತ್ತು ಇತರೆ, ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಉಪಯೋಗಿಸಿ ಕಾರ್ಯವನ್ನು ನಿರ್ವಹಿಸಲು ಸಹಾಯ ನಿರ್ವಹಿಸುತ್ತವೆ.

ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಇಂತಹ ಅನುಭವವನ್ನು ಸೇರಿಸಲು, ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಗಳು ಮತ್ತು ಕೆಲಸ ಮಾಡಬೇಕಾದ ಗ್ರಂಥಾಲಯವನ್ನು ಆಯ್ಕೆಮಾಡುವಂತಹ ಸಂಪ್ರದಾಯಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು. ಇದೇನು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ನೀವು ಕಲಿಯಬಹುದು.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನೀವು:

- openai ಗ್ರಂಥಾಲಯ ಮತ್ತು ಅದರ ಹಿರಿತ ಸಂಪ್ರದಾಯಗಳ ಬಗ್ಗೆ ತಿಳಿದುಕೊಳ್ಳುವಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವಿರಿ.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಹಾಗೂ ಟೋಕನ್ಗಳಂತಹ ಸಂಪ್ರದಾಯಗಳನ್ನು ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವ ವಿಧಾನವನ್ನು ತಿಳಿದುಕೊಳ್ಳುವಿರಿ.

## ಕಲಿಕಾ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು ಮಾಡಲು ಸಾಧ್ಯ:

- ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂಬುದನ್ನು ವಿವರಿಸುವುದು.
- openai ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಹೆಚ್ಚು ಅಥವಾ ಕಡಿಮೆ ಟೋಕನ್‌ಗಳನ್ನು ಉಪಯೋಗಿಸಲು ಕಾನ್ಫಿಗರ್ ಮಾಡುವುದು ಮತ್ತು ತಾಪಮಾನ ಬದಲಾಯಿಸುವ ಮೂಲಕ ವಿಭಿನ್ನಾತ್ಮಕ ಔಟ್ಪುಟ್ ಪಡೆಯುವುದು.

## ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವಾಗ ಈ ಕೆಳಗಿನಂತಹ ಇಂಟರಫೇಸ್ ಇರುವುದನ್ನು ನೋಡಬಹುದು:

- ಕಮಾಂಡ್ ಆಧಾರಿತ. ಕಾನ್ಸೋಲ್ ಅಪ್ಲಿಕೆಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಿದಾಗ ಕಾರ್ಯ ನಿರ್ವಹಿಸುವ ಅಪ್ಲಿಕೇಶನ್‌ಗಳಾಗಿವೆ. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೆಶನ್‌ಗಳಲ್ಲಿ ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇಂಟರಫೇಸ್ (GUIs) ಇರುತ್ತವೆ, ಅಲ್ಲಿ ನೀವು ಬಟನ್‌ಗಳನ್ನು ಕ್ಲಿಕ್ ಮಾಡಬಹುದು, ಪಠ್ಯ ದಾಖಲಿಸಬಹುದು, ಆಯ್ಕೆಯನ್ನು ಮಾಡಬಹುದು ಇವೆ.

### ಕಾನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೆಶನ್‌ಗಳು ಮಿತಿತವಾಗಿವೆ

ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡುವ ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನಿಗೆ ಹೋಲಿಸಿ ಪರಿಗಣಿಸಿ: 

- **ಇದು ಮಿತಿತವಾಗಿದೆ**. ನೀವು ಯಾವುದಾದರೂ ಕಮಾಂಡ್ ಅನ್ನು ಟೈಪ್ ಮಾಡಲಾಗುವುದಿಲ್ಲ, ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಕಮಾಂಡ್‌ಗಳಿಗಷ್ಟೇ ಅವಕಾಶ ಇದೆ.
- **ಭಾಷಾ ವಿಶೇಷ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಹಲವಾರು ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೂ ಅಡಳಿತವಾಗಿ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ವಿಶೇಷ ಭಾಷೆಯ ಹೊರೆಗೆ ನಿರ್ಮಿಸಲಾಗಿದೆ, ಹೆಚ್ಚುವರಿ ಭಾಷಾ ಬೆಂಬಲ ಸೇರಿಸಲು ಸಾಧ್ಯವಿದ್ದರೂ ಸಹ.

### ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ಗಳ ಪ್ರಯೋಜನಗಳು

ಹೀಗಾದರೆ ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೆಶನ್ ಹೇಗೆ ಭಿನ್ನವಾಗಿದೆ?

ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನಿನಲ್ಲಿ, ನೀವು ಹೆಚ್ಚು ಸ್ವಾತಂತ್ರ್ಯ ಹೊಂದಿದ್ದೀರಿ, ನಿರ್ದಿಷ್ಟ ಕಮಾಂಡ್ ಅಥವಾ ಇನ್‌ಪುಟ್ ಭಾಷೆಗೆ ಸೀಮಿತವಲ್ಲ. ಬದಲು, ನೀವು ಸ್ವಾಭಾವಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿಕೊಂಡು ಅಪ್ಲಿಕೇಶನ್ನೊಂದಿಗೆ ಸಂವಹನ ಮಾಡಬಹುದು. ಇನ್ನೊಂದು ಪ್ರಯೋಜನವೆಂದರೆ, ನೀವು ಈಗಾಗಲೇ ವಿಸ್ತೃತ ಮಾಹಿತಿಯ ಸಮೂಹದ ಮೇಲೆ ತರಬೇತುಗೊಂಡಿರುವ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತಿದ್ದೀರಿ, ಆದರೆ ತಾಂಪ್ರಂಪರಿಕ ಅಪ್ಲಿಕೇಶನ್ ಯಾವದು ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿ ಇರುವುದರಲ್ಲಿ ಸೀಮಿತವಾಗಿರಬಹುದು.

### ಪಠ್ಯ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್‌ನೊಂದಿಗೆ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಬಹಳಷ್ಟು ವಸ್ತುಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಬಗ್ಗೆ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಆಯ್ಕೆಯಾಗಬಹುದು.
- **ಸಹಾಯಕ**. LLMಗಳು ಪಠ್ಯ ಸಂಗ್ರಹಣೆ, ಪಠ್ಯದಿಂದ ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು, ರೆಸುಂಬೆಗಳಂತಹ ಪಠ್ಯ ತಯಾರಿಕೆ ಮುಂತಾದ ಕೆಲಸಗಳಲ್ಲಿ ಅತ್ಯುತ್ತಮ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿಸಿ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಪಾಡುಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ನೀವು ಸಾಮಾನ್ಯವಾಗಿ ಎರಡು ವಿಧಾನಗಳನ್ನು ಬಳಸಿಕೊಂಡು LLM ಅಳವಡಿಸುವ ಮಾರ್ಗ ಕಂಡುಹಿಡಿಯಬೇಕು:

- API ಬಳಸುವುದು. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ ತಯಾರಾದ ಪಠ್ಯವನ್ನು ಪಡೆದುಕೊಳ್ಳುತ್ತೀರಿ.
- ಗ್ರಂಥಾಲಯ ಬಳಸುವುದು. ಗ್ರಂಥಾಲಯಗಳು API ಕರೆಗಳನ್ನು ಗುಟ್ಟುಮೈಸು ಮಾಡುತ್ತವೆ ಮತ್ತು ಅವುಗಳನ್ನು ಉಪಯೋಗಿಸಲು ಸುಲಭ ಮಾಡುತ್ತವೆ.

## ಗ್ರಂಥಾಲಯಗಳು/SDKಗಳು

LLMಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡುವ ಕೆಲವು ಪ್ರಸಿದ್ಧ ಗ್ರಂಥಾಲಯಗಳು ಇವೆ:

- **openai**, ಈ ಗ್ರಂಥಾಲಯ ನಿಮ್ಮ ಮಾದರಿಯನ್ನು ಸಂಪರ್ಕಿಸಲು ಮತ್ತು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಕಳುಹಿಸಲು ಸುಲಭ ಮಾಡುತ್ತದೆ.

ನಂತರ, ಮೇಲ್ಮಟ್ಟದಲ್ಲಿ ಕಾರ್ಯನಿರ್ವಹಿಸುವ ಗ್ರಂಥಾಲಯಗಳಿವೆ:

- **Langchain**. Langchain ಪ್ರಸಿದ್ಧವಾಗಿದೆ ಮತ್ತು Python ಅನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Semantic Kernel**. Semantic Kernel ಮೈಕ್ರೋಸಾಫ್ಟ್ ನ ಗ್ರಂಥಾಲಯ, ಇದು C#, Python ಮತ್ತು Java ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.

## openai ಬಳಸಿ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್

ಬನ್ನಿ, ನಾವು ನಮ್ಮ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸಬೇಕೆಂದು, ಯಾವ ಗ್ರಂಥಾಲಯಗಳು ಬೇಕು, ಎಷ್ಟು ಅಗತ್ಯವಿದೆ ಮತ್ತು ಇತರ ವಿಚಾರಗಳನ್ನು ನೋಡೋಣ.

### openai ಇನ್‌ಸ್ಟಾಲ್ ಮಾಡಿ

  > [!NOTE] ಈ ಹಂತವು ನಿಮ್ಮ Codespaces ಅಥವಾ Devcontainer ನಲ್ಲಿ ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು ಸಂಚಾಲಿಸುವಾಗ ಅಗತ್ಯವಿಲ್ಲ


OpenAI ಅಥವಾ Azure OpenAI ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಲು ಹಲವಾರು ಗ್ರಂಥಾಲಯಗಳು ಇವೆ. ನೀವು C#, Python, JavaScript, Java ಮುಂತಾದ ಅನೇಕ ಪ್ರೋಗ್ರಾಮಿಂಗ್ ಭಾಷೆಗಳನ್ನೂ ಬಳಸಬಹುದು.  
ನಾವು `openai` Python ಗ್ರಂಥಾಲಯವನ್ನು ಆರಿಸಿಕೊಂಡಿದ್ದೇವೆ, ಆದ್ದರಿಂದ ನಾವು `pip` ಬಳಸಿ ಇದನ್ನು ಇನ್‌ಸ್ಟಾಲ್ ಮಾಡೋಣ.

```bash
pip install openai
```

ನೀವು Codespaces ಅಥವಾ Devcontainer ನಲ್ಲಿ ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು ಸಂಚಾಲಿಸುತ್ತಿಲ್ಲವಾದರೆ, ನಿಮ್ಮ ಯಂತ್ರದಲ್ಲಿ [Python](https://www.python.org/) ಅನ್ನು ಇನ್‌ಸ್ಟಾಲ್ ಮಾಡಬೇಕು.

> - ಪರಿಸರ ಚರ `AZURE_OPENAI_API_KEY` ಅನ್ನು ನಿಮ್ಮ .env ಫೈಲ್‌ನಲ್ಲಿ ನಿಮ್ಮ API ಕೀಗೆ ಸೆಟ್ ಮಾಡಿ. ನೀವು ಈ ಕೋರ್ಸಿನ ಹಿಂದಿನ ಅಭ್ಯಾಸಗಳನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ್ದರೆ, ಎಲ್ಲವೂ ಸಿದ್ಧವಾಗಿದೆ.


### Azure ಕಾನ್ಫಿಗರೇಶನ್ ಸೆಟ್ ಅಪ್ ಮಾಡಿ

ನೀವು Azure OpenAI ಬಳಿಸುತ್ತಿದ್ದರೆ, ಕೆಳಗಿನ ರೀತಿಯಲ್ಲಿ ಕಾನ್ಫಿಗರೇಶನ್ ಮಾಡಬಹುದು. Responses API ಅನ್ನು Azure OpenAI **v1 ಎಂಡ್‌ಪಾಯಿಂಟ್** ನಿಂದ ಸೇವೆ ನೀಡಿ, ಆದ್ದರಿಂದ `OpenAI` ಕ್ಲೈಂಟ್ ಅನ್ನು `<your-endpoint>/openai/v1/` ಗೆ ಸೂಚಿಸುತ್ತೇವೆ:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ಮೇಲಿನ ಕೋಡಿನಲ್ಲಿ ನಾವು ಈ ಕೆಳಗಿನವರನ್ನು ಸೆಟ್ ಮಾಡುತ್ತಿದ್ದೇವೆ:

- `api_key`, ಇದು ನಿಮ್ಮ API ಕೀ ಆಗಿದ್ದು, Azure ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ಕಂಡುಬರುತ್ತೆ.
- `base_url`, ಇದು ನಿಮ್ಮ Azure OpenAI ಎಂಡ್‌ಪಾಯಿಂಟ್ ಆಗಿದ್ದು, `/openai/v1/` ಸೇರಿಸಲಾಗಿದೆ. ನೀವು API ಕೀಯನ ಪಕ್ಕದಲ್ಲಿ Azure ಪೋರ್ಟಲ್ ನಲ್ಲಿ ಈ ಎಂಡ್‌ಪಾಯಿಂಟ್ ಅನ್ನು ಕಂಡುಹೇಳಬಹುದು. v1 ಎಂಡ್ಡポイント ಬಳಕೆ `api_version` ಪಾಸ್ ಮಾಡಬೇಕಿಲ್ಲ.
- `deployment`, ನೀವು Foundry ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ರಚಿಸಿದ ಮಾದರಿ ನಿಯೋಜನೆಯ ಹೆಸರು.

> [!NOTE]
> `os.environ` ಒಂದು ಮ್ಯಾಪಿಂಗ್ ಆಗಿದ್ದು ಪರಿಸರ ಚರಗಳನ್ನು ಓದುತ್ತದೆ. ನೀವು ಇದನ್ನು `AZURE_OPENAI_API_KEY` ಮತ್ತು `AZURE_OPENAI_ENDPOINT` ನಂತಹ ಪರಿಸರ ಚರಗಳನ್ನು ಓದಲು ಬಳಸಿ.

## ಪಠ್ಯ ತಯಾರಿಕೆ

### ಚಾಟ್ ಪ್ರತಿಕ್ರಿಯೆಗಳು

Responses APIವು ಒಂದೇ ಬಾರಿಗೆ ಪಠ್ಯ ತಯಾರಿಕೆ ಮತ್ತು ಬಹು ಬಾರಿ ಸಂಭಾಷಣೆ ನಡೆಸುವ ಚಾಟ್‌ಬಾಟ್‌ಗಳಿಗೂ ಸೂಕ್ತವಾಗಿದೆ - ನೀವು `input` ಪಟ್ಟಿಯಲ್ಲಿ ಹೆಚ್ಚುವರಿ ಸಂದೇಶಗಳನ್ನು ಒದಗಿಸುವ ಮೂಲಕ ಸಂಭಾಷಣೆಯನ್ನು నిర్మಿಸಬಹುದು:

## ವಿವಿಧ ವಸ್ತುಗಳಿಗೆ ವಿಭಿನ್ನ ರೀತಿಯ ಪ್ರಾಂಪ್ಸ್‍ಗಳು  

ಈಗ ನೀವು ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಪಠ್ಯವನ್ನು ಹೇಗೆ ರಚಿಸಲು ಆಗುತ್ತದೆಯೋ ನೋಡಿದ್ದೀರಿ. ನೀವು ಒಂದು ಪ್ರೋಗ್ರಾಂ ಅನ್ನು ಕಾರ್ಯಾಚರಣೆಗೊಳಿಸಿದ್ದು, ಅದನ್ನು ಬದಲಾಯಿಸಿ ವಿಭಿನ್ನ ರೀತಿಯ ಪಠ್ಯಗಳನ್ನು ರಚಿಸಬಹುದು.  

ಪ್ರಾಂಪ್ಟ್‍ಗಳನ್ನು ಬಗೆಬಗೆಯ ಕಾರ್ಯಗಳಿಗೆ ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ:  

- **ಒಂದು ರೀತಿಯ ಪಠ್ಯ ರಚಿಸಿ**. ಉದಾಹರಣೆಗೆ, ನೀವು ಒಂದು ಕವಿತೆ, ಪ್ರಶ್ನೋತ್ತರಗಳು ಮುಂತಾದವುಗಳನ್ನು ರಚಿಸಬಹುದು.  
- **ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಿ**. ಉದಾಹರಣೆಗೆ 'ವೆಬ್ ಡೆವಲप್ಮೆಂಟ್‌ನಲ್ಲಿ CORS ಎಂದರೇನು?' ಎಂಬ ಪ್ರಶ್ನೆಗೆ ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಲು ಪ್ರಾಂಪ್ಟ್ ಬಳಸಬಹುದು.
