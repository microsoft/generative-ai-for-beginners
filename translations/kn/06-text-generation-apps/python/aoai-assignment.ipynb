# ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ನೀವು ಈ ಪಠ್ಯಕ್ರಮದ ಮೂಲಕ ಈಗಾಗಲೇ ನೋಡಿದ್ದೀರಿ, ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಎಂಜಿನಿಯರಿಂಗ್" ಎಂಬ ಸಂಪೂರ್ಣ ಶಿಸ್ತಿನಂತಹ ಮೂಲಭೂತ ಕಲ್ಪನೆಗಳಿವೆ. ನೀವು ಸಂವಹನ ಮಾಡಬಹುದಾದ ಅನೇಕ ಸಾಧನಗಳು, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮತ್ತು ಇನ್ನಷ್ಟು, ಏನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸುವುದನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ.

ನೀವು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಇಂತಹ ಅನುಭವವನ್ನು ಸೇರಿಸಲು, ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಸುವಿಕೆಗಳು ಮತ್ತು ಕೆಲಸ ಮಾಡಲು ಲೈಬ್ರರಿ ಆಯ್ಕೆಮಾಡುವುದು ಎಂಬ ಕಲ್ಪನೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕಾಗುತ್ತದೆ. ಇದೇನೂ ನೀವು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ಕಲಿಯಲಿದ್ದೀರಿ.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನೀವು:

- openai ಲೈಬ್ರರಿ ಮತ್ತು ಅದರ ಮೂಲಭೂತ ಕಲ್ಪನೆಗಳ ಬಗ್ಗೆ ತಿಳಿಯಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟೋಕನ್‌ಗಳಂತಹ ಕಲ್ಪನೆಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವ ವಿಧಾನವನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಿ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು:

- ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂದು ವಿವರಿಸಬಹುದು.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಬಹುದು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ಹೆಚ್ಚು ಅಥವಾ ಕಡಿಮೆ ಟೋಕನ್‌ಗಳನ್ನು ಬಳಸಲು ಮತ್ತು ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸಲು ಸಂರಚಿಸಬಹುದು, ವಿಭಿನ್ನ ಔಟ್‌ಪುಟ್‌ಗಾಗಿ.

## ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವಾಗ, ಅದು ಕೆಳಗಿನಂತೆಯಾದ ಇಂಟರ್ಫೇಸ್ ಹೊಂದಿರುತ್ತದೆ:

- ಕಮಾಂಡ್ ಆಧಾರಿತ. ಕಾನ್ಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಿ ಕಾರ್ಯವನ್ನು ನಿರ್ವಹಿಸುವ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್‌ಗಳನ್ನು (GUIs) ಹೊಂದಿವೆ, ಅಲ್ಲಿ ನೀವು ಬಟನ್‌ಗಳನ್ನು ಕ್ಲಿಕ್ ಮಾಡಬಹುದು, ಪಠ್ಯವನ್ನು ನಮೂದಿಸಬಹುದು, ಆಯ್ಕೆಯನ್ನು ಆರಿಸಬಹುದು ಮತ್ತು ಇನ್ನಷ್ಟು.

### ಕಾನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸೀಮಿತ

ನೀವು ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡುವ ಕಮಾಂಡ್ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್‌ಗಳೊಂದಿಗೆ ಹೋಲಿಸಿ:

- **ಇದು ಸೀಮಿತವಾಗಿದೆ**. ನೀವು ಯಾವುದೇ ಕಮಾಂಡ್ ಟೈಪ್ ಮಾಡಲಾಗುವುದಿಲ್ಲ, ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಕಮಾಂಡ್‌ಗಳಷ್ಟೇ.
- **ಭಾಷಾ ನಿರ್ದಿಷ್ಟ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಅನೇಕ ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಡೀಫಾಲ್ಟ್‌ನಲ್ಲಿ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗೆ ನಿರ್ಮಿಸಲಾಗಿದೆ, ನೀವು ಹೆಚ್ಚಿನ ಭಾಷಾ ಬೆಂಬಲವನ್ನು ಸೇರಿಸಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಲಾಭಗಳು

ಹೀಗಾಗಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ವಿಭಿನ್ನ?

ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ, ನಿಮಗೆ ಹೆಚ್ಚು ಸ್ವಾತಂತ್ರ್ಯವಿದೆ, ನೀವು ಕಮಾಂಡ್‌ಗಳ ಅಥವಾ ನಿರ್ದಿಷ್ಟ ಇನ್‌ಪುಟ್ ಭಾಷೆಯ ಸೀಮಿತಕ್ಕೆ ಒಳಪಟ್ಟಿಲ್ಲ. ಬದಲಾಗಿ, ನೀವು ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಅಪ್ಲಿಕೇಶನ್ ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಬಹುದು. ಮತ್ತೊಂದು ಲಾಭವೆಂದರೆ, ನೀವು ಈಗಾಗಲೇ ವಿಶಾಲ ಮಾಹಿತಿಯ ಸಂಗ್ರಹದ ಮೇಲೆ ತರಬೇತಿಗೊಂಡಿರುವ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತಿದ್ದೀರಿ, ಆದರೆ ಸಾಂಪ್ರದಾಯಿಕ ಅಪ್ಲಿಕೇಶನ್ ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿರುವ ಮಾಹಿತಿಗೆ ಸೀಮಿತವಾಗಿರಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ನೊಂದಿಗೆ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಅನೇಕ ವಸ್ತುಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಬಗ್ಗೆ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಆಯ್ಕೆ ಆಗಬಹುದು.
- **ಸಹಾಯಕ**. LLMಗಳು ಪಠ್ಯ ಸಾರಾಂಶ ಮಾಡುವುದು, ಪಠ್ಯದಿಂದ ಒಳನೋಟಗಳನ್ನು ಪಡೆಯುವುದು, ರೆಸ್ಯೂಮ್‌ಗಳಂತಹ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವುದು ಮತ್ತು ಇನ್ನಷ್ಟು ಕಾರ್ಯಗಳಲ್ಲಿ ಉತ್ತಮವಾಗಿವೆ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿಸಿ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಅನ್ನು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡಲು ಬಳಸಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ನೀವು ಸಾಮಾನ್ಯವಾಗಿ ಕೆಳಗಿನ ಎರಡು ವಿಧಾನಗಳನ್ನು ಒಳಗೊಂಡಿರುವ LLM ಜೊತೆಗೆ ಏಕೀಕರಣದ ಮಾರ್ಗವನ್ನು ಹುಡುಕಬೇಕಾಗುತ್ತದೆ:

- API ಬಳಸಿ. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ನೊಂದಿಗೆ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ, ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಲೈಬ್ರರಿ ಬಳಸಿ. ಲೈಬ್ರರಿಗಳು API ಕರೆಗಳನ್ನು ಸಂಗ್ರಹಿಸಿ ಅವುಗಳನ್ನು ಬಳಸಲು ಸುಲಭವಾಗಿಸುತ್ತವೆ.

## ಲೈಬ್ರರಿಗಳು/SDKಗಳು

LLMಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು ಕೆಲವು ಪ್ರಸಿದ್ಧ ಲೈಬ್ರರಿಗಳು ಇವೆ:

- **openai**, ಈ ಲೈಬ್ರರಿ ನಿಮ್ಮ ಮಾದರಿಯೊಂದಿಗೆ ಸಂಪರ್ಕ ಸಾಧಿಸಲು ಮತ್ತು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಕಳುಹಿಸಲು ಸುಲಭವಾಗಿಸುತ್ತದೆ.

ಮೇಲ್ಮಟ್ಟದಲ್ಲಿ ಕಾರ್ಯನಿರ್ವಹಿಸುವ ಲೈಬ್ರರಿಗಳು:

- **Langchain**. Langchain ಪ್ರಸಿದ್ಧವಾಗಿದೆ ಮತ್ತು Python ಅನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Semantic Kernel**. Semantic Kernel ಮೈಕ್ರೋಸಾಫ್ಟ್‌ನ ಲೈಬ್ರರಿ, ಇದು C#, Python ಮತ್ತು Java ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.

## openai ಬಳಸಿ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್

ನಾವು ನಮ್ಮ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸಬಹುದು, ಯಾವ ಲೈಬ್ರರಿಗಳು ಬೇಕು, ಎಷ್ಟು ಅಗತ್ಯವಿದೆ ಮತ್ತು ಇತ್ಯಾದಿ ನೋಡೋಣ.

### openai ಅನ್ನು ಸ್ಥಾಪಿಸಿ

  > [!NOTE] ನೀವು ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು Codespaces ಅಥವಾ Devcontainer ನಲ್ಲಿ ಓಡಿಸುತ್ತಿದ್ದರೆ ಈ ಹಂತ ಅಗತ್ಯವಿಲ್ಲ


OpenAI ಅಥವಾ Azure OpenAI ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಲು ಅನೇಕ ಲೈಬ್ರರಿಗಳು ಇವೆ. C#, Python, JavaScript, Java ಮತ್ತು ಇನ್ನಷ್ಟು ಪ್ರೋಗ್ರಾಮಿಂಗ್ ಭಾಷೆಗಳನ್ನು ಬಳಸಬಹುದು.  
ನಾವು `openai` Python ಲೈಬ್ರರಿಯನ್ನು ಬಳಸಲು ಆಯ್ಕೆಮಾಡಿದ್ದೇವೆ, ಆದ್ದರಿಂದ ಅದನ್ನು ಸ್ಥಾಪಿಸಲು `pip` ಬಳಸುತ್ತೇವೆ.

```bash
pip install openai
```

ನೀವು ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು Codespaces ಅಥವಾ Dev Container ನಲ್ಲಿ ಓಡಿಸುತ್ತಿಲ್ಲದಿದ್ದರೆ, ನಿಮ್ಮ ಯಂತ್ರದಲ್ಲಿ [Python](https://www.python.org/) ಅನ್ನು ಸ್ಥಾಪಿಸಬೇಕಾಗುತ್ತದೆ.

### ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸಿ

ನೀವು ಈಗಾಗಲೇ ಮಾಡದಿದ್ದರೆ, ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸಿ:

- Azure ನಲ್ಲಿ ಖಾತೆಯನ್ನು ರಚಿಸಿ <https://azure.microsoft.com/free/>.
- Azure OpenAI ಗೆ ಪ್ರವೇಶ ಪಡೆಯಿರಿ. <https://learn.microsoft.com/azure/ai-services/openai/overview#how-do-i-get-access-to-azure-openai> ಗೆ ಹೋಗಿ ಪ್ರವೇಶಕ್ಕಾಗಿ ವಿನಂತಿ ಮಾಡಿ.

  > [!NOTE]
  > ಬರೆಯುವ ಸಮಯದಲ್ಲಿ, ನೀವು Azure OpenAI ಗೆ ಪ್ರವೇಶಕ್ಕಾಗಿ ಅರ್ಜಿ ಸಲ್ಲಿಸಬೇಕಾಗುತ್ತದೆ.

- Azure OpenAI ಸೇವಾ ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸಿ. [ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸುವ ವಿಧಾನ](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ಈ ಮಾರ್ಗದರ್ಶಿಯನ್ನು ನೋಡಿ.

```python
client = AzureOpenAI(
```

- `azure_endpoint`, ಇದು API ಎಂಡ್ಪಾಯಿಂಟ್. ನೀವು ಅದನ್ನು Azure ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ನಿಮ್ಮ API ಕೀ ಪಕ್ಕದಲ್ಲಿ ಕಾಣಬಹುದು.

## ವಿಭಿನ್ನ ಕಾರ್ಯಗಳಿಗೆ ವಿಭಿನ್ನ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ಪ್ರಕಾರಗಳು

ನೀವು ಈಗ ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಪಠ್ಯವನ್ನು ಹೇಗೆ ರಚಿಸುವುದನ್ನು ನೋಡಿದ್ದೀರಿ. ನೀವು ಬದಲಾಯಿಸಬಹುದಾದ ಮತ್ತು ವಿಭಿನ್ನ ರೀತಿಯ ಪಠ್ಯವನ್ನು ರಚಿಸಲು ಬಳಸಬಹುದಾದ ಒಂದು ಪ್ರೋಗ್ರಾಮ್ ಕೂಡ ಹೊಂದಿದ್ದೀರಿ.

ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಎಲ್ಲಾ ರೀತಿಯ ಕಾರ್ಯಗಳಿಗೆ ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಒಂದು ರೀತಿಯ ಪಠ್ಯವನ್ನು ರಚಿಸಿ**. ಉದಾಹರಣೆಗೆ, ನೀವು ಒಂದು ಕವಿತೆ, ಪ್ರಶ್ನೋತ್ತರಗಳಿಗಾಗಿ ಪ್ರಶ್ನೆಗಳು ಇತ್ಯಾದಿ ರಚಿಸಬಹುದು.
- **ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಿ**. ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಕೆಳಗಿನ ಉದಾಹರಣೆಯಂತೆ 'ವೆಬ್ ಡೆವಲಪ್‌ಮೆಂಟ್‌ನಲ್ಲಿ CORS ಎಂದರೇನು?' ಎಂಬ ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಬಹುದು.
- **ಕೋಡ್ ರಚಿಸಿ**. ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಕೋಡ್ ರಚಿಸಬಹುದು, ಉದಾಹರಣೆಗೆ ಇಮೇಲ್‌ಗಳನ್ನು ಪರಿಶೀಲಿಸಲು ಬಳಸುವ ನಿಯಮಿತ ಅಭಿವ್ಯಕ್ತಿ (regular expression) ರಚಿಸುವುದು ಅಥವಾ ಸಂಪೂರ್ಣ ಪ್ರೋಗ್ರಾಮ್, ಉದಾಹರಣೆಗೆ ವೆಬ್ ಅಪ್ಲಿಕೇಶನ್ ರಚಿಸುವುದನ್ನು ಮಾಡಬಹುದು.

1. ಇತ್ತೀಚೆಗೆ ಇರುವ *app.py* ಫೈಲನ್ನು ಪ್ರಾರಂಭಿಕ ಬಿಂದುವಾಗಿ ಬಳಸಿ
1. `prompt` ಚರವನ್ನು ಹುಡುಕಿ ಮತ್ತು ಅದರ ಕೋಡ್ ಅನ್ನು ಕೆಳಗಿನಂತೆ ಬದಲಾಯಿಸಿ:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

- **ತಾಪಮಾನ (temperature) ಜೊತೆ ಪ್ರಯೋಗ**. ತಾಪಮಾನವು ನಾವು ಇದುವರೆಗೆ ಉಲ್ಲೇಖಿಸದಿದ್ದರೂ, ನಮ್ಮ ಪ್ರೋಗ್ರಾಂ ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ ಎಂಬುದಕ್ಕೆ ಪ್ರಮುಖ ಸಾಂದರ್ಭಿಕವಾಗಿದೆ. ತಾಪಮಾನ ಮೌಲ್ಯ ಹೆಚ್ಚಾದಂತೆ ಔಟ್‌ಪುಟ್ ಹೆಚ್ಚು ಯಾದೃಚ್ಛಿಕವಾಗುತ್ತದೆ. ತಾಪಮಾನ ಮೌಲ್ಯ ಕಡಿಮೆ ಆದಂತೆ ಔಟ್‌ಪುಟ್ ಹೆಚ್ಚು ನಿರೀಕ್ಷಿತವಾಗಿರುತ್ತದೆ. ನೀವು ನಿಮ್ಮ ಔಟ್‌ಪುಟ್‌ನಲ್ಲಿ ವೈವಿಧ್ಯತೆಯನ್ನು ಬಯಸುತ್ತೀರಾ ಅಥವಾ ಇಲ್ಲವೇ ಎಂದು ಪರಿಗಣಿಸಿ.

   ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸಲು, ನೀವು `temperature` ಪರಿಮಾಣವನ್ನು ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ನೀವು 0.5 ತಾಪಮಾನವನ್ನು ಬಳಸಲು ಬಯಸಿದರೆ, ನೀವು ಹೀಗೆ ಮಾಡುತ್ತೀರಿ:

