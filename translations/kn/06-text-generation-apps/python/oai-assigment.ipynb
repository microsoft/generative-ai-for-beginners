# ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ನೀವು ಈ ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ಇದುವರೆಗೆ ನೋಡಿದ್ದು ಪ್ರಾಂಪ್ಟ್‌ಗಳಂತೆ ಮೂಲ ಭಾವನೆಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಎಂಜಿನಿಯರಿಂಗ್" ಎಂಬ ಸಂಪೂರ್ಣ ಶಿಸ್ತಿನಿರುವುದು. ನೀವು ಸಂವಹನ ಮಾಡಬಲ್ಲ ಹಲವು ಟೂಲ್‌ಗಳು ಇದ್ದು, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮತ್ತು ಇನ್ನಷ್ಟು, ನೀವು ಏನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಸುವ ಮೂಲಕ ನಿಮ್ಮನ್ನು ಸಹಾಯ ಮಾಡುತ್ತವೆ.

ನೀವು ಒಂದು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಇಂತಹ ಅನುಭವವನ್ನು ಸೇರಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರ್ಣಗೊಳಿಕೆಗಳು ಎಂಬಂತಹ ಭಾವನೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು ಮತ್ತು ಕೆಲಸ ಮಾಡಲು ಗ್ರಂಥಾಲಯವನ್ನು ಆಯ್ಕೆ ಮಾಡಿಕೊಳ್ಳಬೇಕು. ಇದು ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ನೀವು ಕಲಿಯುವದಾಗಿರುತ್ತದೆ.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ನೀವು:

- openai ಗ್ರಂಥಾಲಯ ಮತ್ತು ಅದರ ಮೂಲಭೂತ ಭಾವನೆಗಳನ್ನು ಅಧ್ಯಯನ ಮಾಡುತ್ತೀರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟ್ವೋಕನ್ಸ್ ಜೇರೆಂದು ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲು ಹೇಗೆ ಬಳಸುವುದು ಅವಗಾಹಿಸು.

## ಕಲಿಕೆ ಗುರಿಗಳು

ಈ ಪಾಠದ ಅಂತರದಲ್ಲಿ ನೀವು ಯೋಗ್ಯರಾಗಿರುತ್ತೀರಿ:

- ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂಬುದನ್ನು ವಿವರಿಸು.
- openai ಬಳಸಿ ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಉಳಿತಾಯ ಮಾಡಲು ಹೆಚ್ಚು ಅಥವಾ ಕಡಿಮೆ ಟ್ವೋಕನ್ಸ್ ಹಾಗೂ ವಿವಿಧ ತಾಪಮಾನವನ್ನು ಹೊಂದಿಸುವುದರಲ್ಲಿ ಕಾನ್ಫಿಗರ್ ಮಾಡು.

## ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೆ ಏನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಎಡಿಟ್ ಮಾಡುವ ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಈ ರೀತಿಯ ಇಂಟರ್ಫೇಸ್ ಇದ್ದಿರುತ್ತದೆ:

- ಆಜ್ಞೆ ಆಧಾರಿತ. ಕನ್ಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು, ನೀವು ಒಂದು ಆಜ್ಞೆಯನ್ನು ಟೈಪ್ ಮಾಡುವಿರಿ ಮತ್ತು ಅದು ಒಂದು ಕಾರ್ಯವನ್ನು ನೆರವೇರಿಸುತ್ತದೆ. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಆಜ್ಞೆ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇಂಟರ್ಫೇಸ್ (GUI) ಇರುತ್ತದೆ, ಅಲ್ಲಿ ನೀವು ಬಟನ್ ಕ್ಲಿಕ್ ಮಾಡುತ್ತೀರಿ, ಪಠ್ಯ ಸೇರಿಸುತ್ತೀರಿ, ಆಯ್ಕೆಯನ್ನು ಆರಿಸುತ್ತೀರಿ ಮತ್ತಿತರ.

### ಕನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಹೊಂದಿರುವ ಮಿತಿಗಳು

ಆಜ್ಞೆ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್‌ where ನೀವು ಒಂದು ಆಜ್ಞೆಯನ್ನು ಟೈಪ್ ಮಾಡುತ್ತೀರಿ ಅಂದರೆ:

- **ಇದು ಮಿತಿಗೊಳಿಸಲಾಗಿದೆ**. ನೀವು ಯಾವುದೇ ಆಜ್ಞೆಯನ್ನು ನಮೂದಿಸಲು ಸಾಧ್ಯವಿಲ್ಲ, ಆ ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಆಜ್ಞೆಗಳನ್ನಷ್ಟೇ ಬಳಸಿ.
- **ಭಾಷಾ ವಿಶೇಷ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಹಲವಾರು ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಮೂಲತಃ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗೆ ನಿರ್ಮಿಸಲಾಗಿದ್ದು ನೀವು ಇನ್ನಷ್ಟು ಭಾಷಾ ಬೆಂಬಲ ಸೇರಿಸಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಪ್ರಯೋಜನಗಳು

ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ವಿಭಿನ್ನವಾಗಿದೆ?

ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ನೀವು ಹೆಚ್ಚು ಸ್ವಾತಂತ್ರ್ಯ ಹೊಂದಿದ್ದೀರಿ, ನೀವು ಒಂದು ನಿರ್ದಿಷ್ಟ ಆಜ್ಞೆಗಳ ಸಮೂಹ ಅಥವಾ ನಿರ್ದಿಷ್ಟ ಇನ್ಪುಟ್ ಭಾಷೆಗೆ ಮಿತಿಯಾಗಿರುವುದಿಲ್ಲ. ಬದಲಾಗಿ, ನೀವು ಸಹಜ ಭಾಷೆಯನ್ನು ಬಳಸಿಕೊಂಡು ಅಪ್ಲಿಕೇಶನ್ ಜೊತೆ ಸಂವಹನ ಮಾಡಬಹುದು. ಇನ್ನೊಂದು ಪ್ರಯೋಜನವೆಂದರೆ ನೀವು ಈಗಾಗಲೇ ದೊಡ್ಡ ಮಾಹಿತಿಯ ಸಂಗ್ರಹದ ಮೇಲೆ ತರಬೇತಿ ಪಡೆಯುವ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತಿದ್ದೀರಿ, ಆದರೆ ಪರंपರাগত ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ಡೇಟಾಬೇಸಿನಲ್ಲಿರುವ ವಿಷಯಗಳಿಗೆ ಮಾತ್ರ ಮಿತಿಯಾಗಿರಬಹುದು.

### ಪಠ್ಯ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್‌ ಮೂಲಕ ನಾನು ಏನು ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಅನೇಕ ವಸ್ತುಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **一个 ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಕುರಿತು ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಒಳ್ಳೆಯ ಆಯ್ಕೆಯಾಗಬಹುದು.
- **ಸಹಾಯಕರ**. LLM ಗಳು ಪಠ್ಯ ಸಾರಾಂಶ ಕೊಡುವುದು, ಪಠ್ಯದೊಳಗಿನ ತತ್ವಗಳನ್ನು ಪಡೆಯುವುದು, ಜೀವನಚರಿತ್ರೆಗಳಂತಹ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವುದರಲ್ಲಿ ನಿಪುಣರು.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿತವಾಗಿಯೇ ನೀವು ಒಂದು ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು, ಇದು ಕೋಡ್ ಬರೆದಿಕೊಳ್ಳಲು ಸಹಾಯಮಾಡುತ್ತದೆ. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಅನ್ನು ಬಳಸಿ ಕೋಡ್ ಬರೆಯಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ಉತ್ತಮವಾಗಿ, ನೀವು LLM ಜೊತೆ ಏಕೆಂದರೆ ಸಂಯೋಜಿಸುವ ಮಾರ್ಗವನ್ನು ಕಂಡುಹಿಡಿಯಬೇಕಾಗುತ್ತದೆ, ಸಾಮಾನ್ಯವಾಗಿ ಕೆಳಗಿನ ಎರಡು ವಿಧಾನಗಳನ್ನು ಒಳಗೊಂಡಿರುತ್ತದೆ:

- API ಬಳಸಿರಿ. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಜೊತೆ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ, ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಗ್ರಂಥಾಲಯ (ಲೈಬ್ರರಿ) ಬಳಸಿ. ಗ್ರಂಥಾಲಯಗಳು API ಕಾಲ್‌ಗಳನ್ನು ಸ್ಲೀಕರಿಸಿ ಬಲವಾಗಿ ಬಳಸಲು ಸುಲಭಗೊಳಿಸುತ್ತವೆ.

## ಗ್ರಂಥಾಲಯಗಳು/SDK ಗಳು

LLMಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು ಕೆಲ ಪ್ರಸಿದ್ಧ ಗ್ರಂಥಾಲಯಗಳು ಇವೆ:

- **openai**, ಈ ಗ್ರಂಥಾಲಯ ನಿಮ್ಮ ಮಾದರಿಯೊಂದಿಗೆ ಸಂಪರ್ಕ ಸ್ಥಾಪಿಸಿ ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಕಳುಹಿಸಲು ಸುಲಭಗೊಳಿಸುತ್ತದೆ.

ನಂತರ ಕೆಲವು ಮೇಲ್ಹೊಂದಿಗಿರುವ ಲೈಬ್ರರಿಗಳು ಇವೆ:

- **Langchain**. Langchain ಪ್ರಸಿದ್ಧವಾಗಿದ್ದು Python ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Semantic Kernel**. Semantic Kernel ಒಂದು Microsoft ಲೈಬ್ರರಿ, ಇದು C#, Python, ಮತ್ತು Java ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.

> ನಿಮ್ಮ API ಕೀ ಮತ್ತು ನಿಮ್ಮ ಕೋಡ್ ಅನ್ನು ಪ್ರತ್ಯೇಕಿಸುವುದು ಒಳಿತು. ನೀವು ಇದನ್ನು ಪರಿಸರ ಚರಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಮಾಡಬಹುದು.

> - API ಕೀ ಕೇವಲ ಒಂದು ಬಾರಿ ಪ್ರವೇಶ ಮಾಡಬಹುದು. ಆದ್ದರಿಂದ ಸರಿಯಾಗಿ ನಕಲಿ ಮಾಡಲಾಗಿದೆ ಎಂಬುದನ್ನು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಿ. ಅದು ಕೆಲಸ ಮಾಡದಿದ್ದರೆ, ಕೀ ಅನ್ನು ಅಳಿಸಿ ಹೊಸ ಕೀ ರಚಿಸುವುದು ಸೂಕ್ತ.


### OpenAI ಸಂರಚನೆಯನ್ನು ಸೆಟ್ ಮಾಡುವುದು

ನೀವು OpenAI ಬಳಸುತ್ತಿದ್ದರೆ, ಈ ರೀತಿ ಸಮರಚನೆ ಮಾಡಬಹುದು:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

ಮೇಲಿನ ಕೋಡ್‌ನಲ್ಲಿ ನಾವು ಕೆಳಕಂಡುವುಗಳನ್ನು ಹೊಂದಿಸುತ್ತೇವೆ:

- `api_key`, ಇದು OpenAI ಡ್ಯಾಶ್‌ಬೋರ್ಡ್‌ನಲ್ಲಿ ಸಿಗುವ ನಿಮ್ಮ API ಕೀ.
- `deployment`, ಇದು ನಿಮ್ಮ GPT ಆವೃತ್ತಿ.

> [!NOTE]
> `os.environ` ಒಂದು ಕಾರ್ಯವಾಗಿದೆ, ಇದು ಪರಿಸರ ಚರಗಳನ್ನು ಓದುತ್ತದೆ. ನೀವು ಇದನ್ನು `OPENAI_API_KEY` ರೀತಿಯ ಪರಿಸರ ಚರಗಳನ್ನು ಓದುವಲ್ಲಿ ಬಳಸಬಹುದು.

## ಪಠ್ಯವು ಉತ್ಪಾದನೆ

ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವ ವಿಧಾನವೆಂದರೆ `responses.create` ಮೆತ್‌ಡ್ ಬಳಕೆಯಾಗುವುದು. ಇಲ್ಲಿದೆ ಒಂದು ಉದಾಹರಣೆ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ಮೇಲಿನ ಕೋಡ್‌ನಲ್ಲಿ ನಾವು ಪ್ರತಿಕ್ರಿಯೆ ವಸ್ತುವನ್ನು ರಚಿಸಿ, ನಾವು ಬಳಸಲು ಬಯಸುವ ಮಾದರಿಯನ್ನು ಮತ್ತು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಹಂಚುತ್ತೇವೆ. ನಂತರ ನಾವು ಉತ್ಪಾದಿತ ಪಠ್ಯವನ್ನು ಮುದ್ರಿಸುತ್ತೇವೆ.

### ಚಾಟ್ ಪ್ರತಿಕ್ರಿಯೆಗಳು

ಇದುವರೆಗೆ, ನೀವು ಒಂದು ಪ್ರಾಂಪ್ಟ್‌ನಿಂದ ಪಠ್ಯವನ್ನು ಉತ್ಪಾದಿಸುವುದನ್ನು ನೋಡಿದ್ದೀರಿ. Responses API ಚಾಟ್‌ಬಾಟ್‌ಗಳಿಗೆ ತುಂಬಾ ಒಪ್ಪಿಕೊಳ್ಳುವದು - ನೀವು ಸಂವಾದ ನಿರ್ಮಿಸಲು `input` ಆಗಿ ಸಂದೇಶಗಳ ಪಟ್ಟಿ ಒದಗಿಸುತ್ತೀರಿ. ಇಲ್ಲಿದೆ ಬಳಸುವುದು ಹೇಗೆ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

ಈ ಕಾರ್ಯಚಟುವಟಿಕೆಯ ಎರಡನ್ನೂ ಮುಂದೆ ಬರುವ ಅಧ್ಯಾಯದಲ್ಲಿ ಹೆಚ್ಚು ವಿವರಿಸಲ್ಪಡುತ್ತದೆ.

## ವಿಭಿನ್ನ ಕೆಲಸಗಳಿಗೆ ವಿಭಿನ್ನ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ವಿಧಗಳು

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

   ತಾಪಮಾನ ಬದಲಾಯಿಸಲು, ನೀವು `temperature` ಪರಿಮಾಣವನ್ನು ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ನೀವು 0.5 ತಾಪಮಾನ ಹಾಗೆಯೇ ಬಳಸಬೇಕಾದರೆ, ನೀವು ಹೀಗೆ ಮಾಡಬಹುದು:
