# ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸಿ

ಈ ಪಾಠಕ್ರಮದ ಮೂಲಕ ನೀವು ಈಗಾಗಲೇ ನೋಡಿರುವಂತೆ, ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಮತ್ತು "ಪ್ರಾಂಪ್ಟ್ ಇಂಜಿನಿಯರಿಂಗ್" ಎಂದು ಕರೆಯಲ್ಪಡುವ ಸಂಪೂರ್ಣ ಶಿಸ್ತಿನಂತಹ ಮೂಲ ಕಲ್ಪನೆಗಳು ಇವೆ. ನೀವು ಸಂವಹನ ಮಾಡಬಹುದಾದ ಅನೇಕ ಸಾಧನಗಳು, ಉದಾಹರಣೆಗೆ ChatGPT, Office 365, Microsoft Power Platform ಮತ್ತು ಮತ್ತಿತರವು, ನೀವು ಏನನ್ನಾದರೂ ಸಾಧಿಸಲು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಬಳಕೆಮಾಡುವಂತೆ ಬೆಂಬಲಿಸುತ್ತವೆ.

ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನಿಗೆ ಇಂತಹ ಅನುಭವವನ್ನು ಸೇರಿಸಲು, ನೀವು ಪ್ರಾಂಪ್ಟ್‌ಗಳು, ಪೂರೈಸುವಿಕೆಗಳಂತಹ ಕಲ್ಪನೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬೇಕು ಮತ್ತು ಕಾರಿಗೊಂದು ಲೈಬ್ರರಿಯನ್ನು ಆಯ್ಕೆಮಾಡಿ ಕೆಲಸಮಾಡಬೇಕಾಗುತ್ತದೆ. ಈ ಅಧ್ಯಾಯದಲ್ಲಿ ನೀವು ಅದೇ ಕಲಿಯುವಿರಿ.

## ಪರಿಚಯ

ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನೀವು:

- openai ಲೈಬ್ರರಿ ಮತ್ತು ಅದರ ಮೂಲ ಕಲ್ಪನೆಗಳ ಬಗ್ಗೆ ಕಲಿಯಿರಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕೆಯ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ.
- ಪ್ರಾಂಪ್ಟ್, ತಾಪಮಾನ ಮತ್ತು ಟೋಕನ್ ಗಳಂತಹ ಕಲ್ಪನೆಗಳನ್ನು ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕೆಯ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ನಿರ್ಮಿಸುವುದೆಂಬುದನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಿ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ, ನೀವು ಮಾಡಬಲ್ಲದು:

- ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು ಎಂದು ವಿವರಿಸಿ.
- openai ಬಳಸಿ ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ನ್ನು ಹೆಚ್ಚು ಅಥವಾ ಕಡಿಮೆ ಟೋಕನ್‌ಗಳನ್ನು ಬಳಸುವಂತೆ, ಹಾಗೂ ತಾಪಮಾನವನ್ನು ಬದಲಾಯಿಸುವಂತೆ ಮಾನ್ಯಗೊಳಿಸಿ, ವಿಭಿನ್ನ ಫಲಿತಾಂಶಕ್ಕಾಗಿ.

## ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ಎಂದರೇನು?

ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿದಾಗ ಅದರ ಕರತಣೆಯಲ್ಲಿ ಹೀಗೆ ಏನೇನೋ ಇರುತ್ತದೆ:

- ಆಜ್ಞೆ ಆಧಾರಿತ. ಕಾನ್ಸೋಲ್ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸಾಮಾನ್ಯವಾಗಿ ನೀವು ಆಜ್ಞೆ ಟೈಪ್ ಮಾಡಿ ಅದನ್ನು ಕಾರ್ಯಗತಗೊಳಿಸುವ ಅಪ್ಲಿಕೇಶನ್ಗಳು. ಉದಾಹರಣೆಗೆ, `git` ಒಂದು ಆಜ್ಞೆ ಆಧಾರಿತ ಅಪ್ಲಿಕೇಶನ್.
- ಬಳಕೆದಾರ ಇನ್‌ಟರ್ಫೇಸ್ (UI). ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಗ್ರಾಫಿಕಲ್ ಬಳಕೆದಾರ ಇನ್‌ಟರ್ಫೇಸ್‌ಗಳಿದ್ದು, ನೀವು ಬಟನ್ ಕ್ಲಿಕ್ ಮಾಡುವುದು, ಪಠ್ಯ ನೀಡುವುದು, ಆಯ್ಕೆಗಳು ಮಾಡುವಂತಹ ಎಲ್ಲಾ ಸಂವಹನ.

### ಕಾನ್ಸೋಲ್ ಮತ್ತು UI ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಸೀಮಿತವಾಗಿವೆ

ನೀವು ಕಾನ್ಸೋಲ್ ಆಜ್ಞೆಯನ್ನು ಟೈಪ್ ಮಾಡುವ ಅಪ್ಲಿಕೇಶನ್ಗೆ ಹೋಲಿಸಿ ನೋಡಿದರೆ:

- **ಇದು ಸೀಮಿತವಾಗಿದೆ**. ನೀವು ಯಾವ ಆಜ್ಞೆ ಬೇಕಾದರೂ ಟೈಪ್ ಮಾಡಲಾಗುವುದಿಲ್ಲ, ಅಪ್ಲಿಕೇಶನ್ ಬೆಂಬಲಿಸುವ ಆಜ್ಞೆಗಳು ಮಾತ್ರ.
- **ಭಾಷೆ ನಿರ್ದಿಷ್ಟ**. ಕೆಲವು ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಅನೇಕ ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತವೆ, ಆದರೆ ಡಿಫಾಲ್ಟ್‌ನಲ್ಲಿ ಅಪ್ಲಿಕೇಶನ್ ಒಂದು ನಿರ್ದಿಷ್ಟ ಭಾಷೆಗೆ ನಿರ್ಮಿಸಲಾಗುತ್ತದೆ, ಹೆಚ್ಚುವರಿ ಭಾಷಾ ಬೆಂಬಲವನ್ನು ಸೇರಿಸಲು ಸಾಧ್ಯವಿದ್ದರೂ.

### ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳ ಪ್ರಯೋಜನಗಳು

ಆದ್ರೆ ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ಹೇಗೆ ವಿಭಿನ್ನವಾಗಿದೆ?

ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ, ನೀವು ಹೆಚ್ಚು свобೊಲತೆ ಹೊಂದಿದ್ದೀರಿ, ನೀವು ನಿಗದಿತ ಆಜ್ಞೆಗಳ ಅಥವಾ ಒಂದು ನಿರ್ದಿಷ್ಟ ಇನ್‌ಪುಟ್ ಭಾಷೆಗೆ ಸೀಮಿತವಾಗಿಲ್ಲ. ಬದಲಾಗಿ, ನಿಮ್ಮನ್ನು ಈ ಅಪ್ಲಿಕೇಶನ್ನೊಂದಿಗೆ ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನೇ ಬಳಸಿ ಸಂವಹನ ಮಾಡಬಹುದು. ಇನ್ನೊಂದು ಪ್ರಯೋಜನವೆಂದರೆ ನೀವು ಈಗಾಗಲೇ ವಿಶಾಲ ಮಾಹಿತಿ ಸಂಗ್ರಹದ ಡೇಟಾ ಮೂಲದೊಂದಿಗೆ ಸಂವಹನ ಮಾಡುತ್ತೀರಿ, ಆಗ ಸಾಮಾನ್ಯ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿ ಇರುತ್ತದೆ ಏನೆಂಬುದರ ಕೈಮುಗಿದು ಇರುತ್ತದೆ.

### ಪಠ್ಯ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್‌ನಿಂದ ನಾನು ಏನನ್ನಾ ನಿರ್ಮಿಸಬಹುದು?

ನೀವು ಹಲವು ವಿಚಾರಗಳನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ:

- **ಚಾಟ್‌ಬಾಟ್**. ನಿಮ್ಮ ಕಂಪನಿ ಮತ್ತು ಅದರ ಉತ್ಪನ್ನಗಳ ಕುರಿತು ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಿಸುವ ಚಾಟ್‌ಬಾಟ್ ಉತ್ತಮ ಆಯ್ಕೆ ಆಗಬಹುದು.
- **ಸಹಾಯಕ**. LLM ಗಳು ಪಠ್ಯ ಸಂಕ್ಷಿಪ್ತಗೊಳಿಸುವಿಕೆ, ಪಠ್ಯದಿಂದ ಅರಿವು ಪಡೆದ ಕ್ರಿಯೆಗಳು, ರೆಸ್ಯೂಮೆಗಳಂತಹ ಪಠ್ಯಗಳನ್ನು ತಯಾರಿಸುವಿಕೆ ಮುಂತಾದಲ್ಲಿ ಉತ್ತಮವಾಗಿವೆ.
- **ಕೋಡ್ ಸಹಾಯಕ**. ನೀವು ಬಳಸುವ ಭಾಷಾ ಮಾದರಿಯ ಮೇಲೆ ಅವಲಂಬಿತವಾಗಿ, ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಮಾಡುವ ಕೋಡ್ ಸಹಾಯಕವನ್ನು ನಿರ್ಮಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, GitHub Copilot ಮತ್ತು ChatGPT ಗಳನ್ನು ಬಳಸಿ ನೀವು ಕೋಡ್ ಬರೆಯಲು ಸಹಾಯ ಪಡೆಯಬಹುದು.

## ನಾನು ಹೇಗೆ ಪ್ರಾರಂಭಿಸಬಹುದು?

ನೀವು ಸಾಮಾನ್ಯವಾಗಿ LLM ಜೊತೆಗೆ ಸಂಯೋಜಿಸಲು ಎರಡು ವಿಧಾನಗಳನ್ನು ಅನುಸರಿಸಬಹುದು:

- API ಬಳಸಿ. ಇಲ್ಲಿ ನೀವು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ಗಳಿಂದ ವೆಬ್ ವಿನಂತಿಗಳನ್ನು ರಚಿಸಿ, ತಯಾರಿಸಲಾದ ಪಠ್ಯವನ್ನು ಪಡೆಯುತ್ತೀರಿ.
- ಲೈಬ್ರರಿ ಬಳಸಿ. ಲೈಬ್ರರಿಗಳು API ಕರೆಗಳನ್ನು ಮುಚ್ಚಿ ಸುಲಭವಾಗಿ ಉಪಯೋಗಿಸುವಂತೆ ನೆರವಾಗುತ್ತವೆ.

## ಲೈಬ್ರರಿಗಳು/SDKಗಳು

LLM ಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು ಕೆಲವು ಪ್ರಸಿದ್ಧ ಲೈಬ್ರರಿಗಳಿವೆ:

- **openai**, ಈ ಲೈಬ್ರರಿ ನಿಮ್ಮ ಮಾದರಿಯ ಜೊತೆಗೆ ಸಂಪರ್ಕ ಹೊಂದುತ್ತದೆ ಮತ್ತು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಕಳುಹಿಸುವುದನ್ನು ಸುಲಭಗೊಳಿಸುತ್ತದೆ.

ಅದಲ್ಲದೆ ಎತ್ತರ ಮಟ್ಟದ ಲೈಬ್ರರಿಗಳೂ ಇವೆ:

- **Langchain**. Langchain ಚೆನ್ನಾಗಿ ಪರಿಚಿತವಾಗಿದೆ ಮತ್ತು Python ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Semantic Kernel**. Semantic Kernel Microsoft ನ ಲೈಬ್ರರಿ, ಇವು C#, Python ಮತ್ತು Java ಭಾಷೆಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.

## openai ಬಳಸಿ ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್

ನಾವು ಮೊದಲ ಅಪ್ಲಿಕೇಶನ್ನನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸಬಹುದು, ಯಾವ ಲೈಬ್ರರಿಗಳು ಬೇಕು, ಎಷ್ಟು ಅಗತ್ಯವಿದೆ ಇವೆಂಬುದನ್ನು ನೋಡೋಣ.

### openai ಸ್ಥಾಪನೆ ಮಾಡಿ

  > [!NOTE] ನೀವು Codespaces ನಲ್ಲಿ ಅಥವಾ Devcontainer ಒಳಗೆ ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು இயக்கುತ್ತಿದ್ದರೆ ಈ ಹಂತ ಅಗತ್ಯವಿಲ್ಲ.


OpenAI ಅಥವಾ Azure OpenAI ಜೊತೆ ಸಂವಹನ ಮಾಡಲು ಅನೇಕ ಲೈಬ್ರರಿಗಳು ಲಭ್ಯವಿವೆ. ನೀವು C#, Python, JavaScript, Java ಮತ್ತು ಇನ್ನಷ್ಟು ಹಲವು ಪ್ರೋಗ್ರಾಮಿಂಗ್ ಭಾಷೆಗಳಲ್ಲೂ ಬಳಸಬಹುದು.  
ನಾವು `openai` Python ಲೈಬ್ರರಿಯನ್ನು ಬಳಸಲು ಆಯ್ಕೆ ಮಾಡಿದ್ದೇವೆ, ಸಹಜವಾಗಿ `pip` ಬಳಸಿ ಇದನ್ನು ಸ್ಥಾಪಿಸುವುದು.

```bash
pip install openai
```

ನೀವು Codespaces ಅಥವಾ Dev Container ನಲ್ಲಿ ಈ ನೋಟ್ಬುಕ್ ಅನ್ನು ಚಲಾಯಿಸುವುದಿಲ್ಲವಾದರೆ, ನಿಮ್ಮ ಯಂತ್ರದಲ್ಲಿ [Python](https://www.python.org/) ಅನ್ನು ಸ್ಥಾಪಿಸಬೇಕಾಗುತ್ತದೆ.

### ಸಂಪನ್ಮೂಲವನ್ನು ರಚಿಸಿ ಮತ್ತು ನಿಮ್ಮ API ಕೀ ಹುಡುಕಿ

ನೀವು ಇನ್ನೂ ಮಾಡದಿದ್ದರೆ, ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸಿ:

- OpenAIಲ್ಲಿ ಖಾತೆ ರಚಿಸಿ <https://platform.openai.com/signup>.
- ಈಗ, ನಿಮ್ಮ API ಕೀಲಿಗಳನ್ನು ಪಡೆದುಕೊಳ್ಳಿ <https://platform.openai.com/api-keys>. 

>[!NOTE]
> ನಿಮ್ಮ API ಕೀಲಿಯನ್ನು ನಿಮ್ಮ ಕೋಡ್‌ನಿಂದ ಪ್ರತ್ಯೇಕಿಸುವುದು ಮುಖ್ಯ. ನೀವು ಇದನ್ನು ಪರಿಸರ ಚರಗಳು ಬಳಸಿ ಮಾಡಬಹುದು.
> - ನಿಮ್ಮ .env ಕಡತದಲ್ಲಿ ಪರಿಸರ ಚರ `OPENAI_KEY` ಅನ್ನು ನಿಮ್ಮ API ಕೀಗೆ ಸೆಟ್ ಮಾಡಿ. ನೀವು ಈಗಾಗಲೇ ಈ ಕೋರ್ಸ್‌ನ ಹಿಂದಿನ ವ್ಯಾಯಾಮಗಳನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ್ದರೆ, ನೀವು ಸಿದ್ಧರಿದ್ದೀರಿ.
> - API ಕೀ ಒಂದು ಬಾರಿ ಮಾತ್ರ ಲಭ್ಯವಿರುತ್ತದೆ ಮತ್ತು ಅದನ್ನು ಸರಿಯಾಗಿ ನಕಲಿಸಿಕೊಂಡಿದ್ದೀರಿ ಎಂದು ಖಚಿತಪಡಿಸಬೇಕು. ಅದು ಕಾರ್ಯನಿರ್ವಹಿಸಲಿಲ್ಲದಿಂದ, ಕೀಅನ್ನು ಅಳಿಸಿ ಹೊಸದನ್ನು ರಚಿಸುವುದು ಶಿಫಾರಸು.


### OpenAI ಸಂರಚನೆಯನ್ನು ಸಿದ್ಧಪಡಿಸಿ

ನೀವು OpenAI ಬಳಸುತ್ತಿದ್ದರೆ, ಸಂರಚನೆ ಸುಳಿವು ಇಲ್ಲಿದೆ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

1. ಇತ್ತೀಚಿನ *app.py* ಕಡತವನ್ನು ಪ್ರಾರಂಭ ಬಿಂದು ಎಂದು ಬಳಸಿ
1. `prompt` ಚರ ಭೇದದವನ್ನು ಕಂಡು ಅದರ ಕೋಡ್ ಅನ್ನು ಕೆಳಗಿನಂತೆ ಬದಲಿ ಮಾಡಿ:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
