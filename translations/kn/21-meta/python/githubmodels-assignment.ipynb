## ಲ್ಲಾಮಾ 3.1 

405 ಬಿಲಿಯನ್ ಪ್ಯಾರಾಮೀಟರ್ಸ್ ಹೊಂದಿರುವ ಲ್ಲಾಮಾ 3.1 ತೆರೆಯಲಾದ ಮೂಲದ LLM ವರ್ಗದಲ್ಲಿ ಸರಿಹೋಗುತ್ತದೆ. 

ಈ ಮೋಡ್ ಹಿಂದಿನ ಬಿಡುಗಡೆಯಾದ ಲ್ಲಾಮಾ 3 ರ ಅಪ್‌ಗ್ರೇಡ್ ಆಗಿದ್ದು: 

- ದೊಡ್ಡ ಸಂದರ್ಭ ಕಿಟಕಿ - 128k ಟೋಕೆನ್ಸ್ ವಿರುದ್ಧ 8k ಟೋಕೆನ್ಸ್ 
- ದೊಡ್ಡ ಗರಿಷ್ಠ ಔಟ್‌ಪುಟ್ ಟೋಕೆನ್ಸ್ - 4096 ವಿರುದ್ಧ 2048 
- ಉತ್ತಮ ಬಹುಭಾಷಾ ಬೆಂಬಲ - ತರಬೇತಿ ಟೋಕೆನ್ಸ್ ಹೆಚ್ಚಳದಿಂದ 

ಇದರಿಂದ ಲ್ಲಾಮಾ 3.1 ಜನಎಐ ಅಪ್ಲಿಕೇಶನ್ಗಳ ನಿರ್ಮಾಣದಲ್ಲಿ ಹೆಚ್ಚು ಸಂಕೀರ್ಣ ಬಳಕೆ ಪ್ರಕರಣಗಳನ್ನು ನಿರ್ವಹಿಸಲು ಸಾಧ್ಯವಾಗುತ್ತದೆ, ಇದರಲ್ಲಿ: 
- ಸ್ಥಳೀಯ ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ - LLM ಕಾರ್ಯಪ್ರವಾಹದ ಹೊರತಾಗಿರುವ ಬಾಹ್ಯ ಉಪಕರಣಗಳು ಮತ್ತು ಫಂಕ್ಷನ್ಗಳನ್ನು ಕರೆ ಮಾಡುವ ಸಾಮರ್ಥ್ಯ
- ಉತ್ತಮ RAG ಕಾರ್ಯಕ್ಷಮತೆ - ಹೆಚ್ಚಿನ ಸಂದರ್ಭ ಕಿಟಕಿಯ ಕಾರಣದಿಂದ 
- ಸಿಂಥೆಟಿಕ್ ಡೇಟಾ ರಚನೆ - ಫೈನ್-ಟ್ಯೂನಿಂಗ್ ಮುಂತಾದ ಕಾರ್ಯಗಳಿಗೆ ಪರಿಣಾಮಕಾರಿಯಾದ ಡೇಟಾವನ್ನು ಸೃಷ್ಟಿಸುವ ಸಾಮರ್ಥ್ಯ 



### ನೇಟಿವ್ ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್

Llama 3.1 ಫಂಕ್ಷನ್ ಅಥವಾ ಉಪಕರಣ ಕಾಲ್‌ಗಳನ್ನು ಹೆಚ್ಚು ಪರಿಣಾಮಕಾರಿಯಾಗಿ ಮಾಡಲು ಉತ್ತಮಗೊಳಿಸಲಾಗಿದೆ. ಇದಲ್ಲದೆ, ಬಳಕೆದಾರನ ಪ್ರಾಂಪ್ಟ್ ಆಧಾರದ ಮೇಲೆ ಬಳಸಬೇಕಾದ ಉಪಕರಣಗಳಾಗಿ ಮಾದರಿಯು ಗುರುತಿಸಬಹುದಾದ ಎರಡು ನಿರ್ಮಿತ ಉಪಕರಣಗಳನ್ನು ಹೊಂದಿದೆ. ಈ ಉಪಕರಣಗಳು ಇವೆ:

- **Brave Search** - ವೆಬ್ ಶೋಧನೆ ನಡೆಸಿ ಹವಾಮಾನತ್ತಿಗಂತಹ ಇತ್ತೀಚಿನ ಮಾಹಿತಿಯನ್ನು ಪಡೆಯಲು ಬಳಸಬಹುದು
- **Wolfram Alpha** - ಸ್ವಂತ ಫಂಕ್ಷನ್‌ಗಳನ್ನು ಬರೆಯಬೇಕಾಗದೆ, ಹೆಚ್ಚು ಸಂಕೀರ್ಣ ಗಣಿತೀಯ ಲೆಕ್ಕಾಚಾರಗಳಿಗೆ ಬಳಸಬಹುದು.

ನೀವು ತಮ್ಮ ಸ್ವಂತ ಕಸ್ಟಮ್ ಉಪಕರಣಗಳನ್ನು ಸೃಷ್ಟಿಸಬಹುದು, ಅವನ್ನು LLM ಕಾಲ್ ಮಾಡಬಹುದು.

ಕೆಳಗಿನ ಕೋಡ್ ಉದಾಹರಣೆಯಲ್ಲಿ:

- ಸಿಸ್ಟಮ್ ಪ್ರಾಂಪ್ಟ್‌ನಲ್ಲಿ ಲಭ್ಯವಿರುವ ಉಪಕರಣಗಳನ್ನು (brave_search, wolfram_alpha) ವಿವರಿಸಲಾಗುತ್ತದೆ.
- ಒಂದು ನಗರದಲ್ಲಿ ಹವಾಮಾನ ಕುರಿತು ಕೇಳುವ ಬಳಕೆದಾರನ ಪ್ರಾಂಪ್ಟ್ ಕಳುಹಿಸಲಾಗುತ್ತದೆ.
- LLM Brave Search ಉಪಕರಣಕ್ಕೆ ಈ ರೀತಿಯ ಕಾಲ್ ನಿರೀಕ್ಷಿಸಬಹುದು `<|python_tag|>brave_search.call(query="Stockholm weather")`

*ಟಿಪ್ಪಣಿ: ಈ ಉದಾಹರಣೆ ಕೇವಲ ಉಪಕರಣ ಕಾಲ್ ಮಾತ್ರ ಮಾಡುತ್ತದೆ, ಫಲಿತಾಂಶಗಳನ್ನು ಪಡೆಯಲು Brave API ಪುಟದಲ್ಲಿ ಉಚಿತ ಖಾತೆಯನ್ನು ರಚಿಸಿ ಮತ್ತು ಫಂಕ್ಷನ್ ಅನ್ನು ನಿರ್ದಿಷ್ಟ ಮಾಡಬೇಕಾಗುತ್ತದೆ*


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### ಲ್ಲಾಮಾ 3.2 

ಲ್ಲಾಮಾ 3.1 ಯೊಂದಿನಾದರೂ ಮಲ್ಟಿಮೋಡಾಲಿಟಿ ಎಂಬ ಒಂದು ಮಿತಿಯಿದೆ, ಆಗಾಗ LLM ಆಗಿದ್ದರೂ ಕೂಡ. ಅಂದರೆ, ಚಿತ್ರಗಳಂತಹ ವಿಭಿನ್ನ ರೀತಿಯ ಇನ್‌ಪುಟ್‌ಗಳನ್ನು ಪ್ರಾಂಪ್ಟ್‌ಗಳಾಗಿ ಬಳಸುವುದು ಮತ್ತು ಉತ್ತರಗಳನ್ನು ನೀಡುವುದು. ಈ ಸಾಮರ್ಥ್ಯ ಲ್ಲಾಮಾ 3.2 ರ ಪ್ರಮುಖ ವೈಶಿಷ್ಟ್ಯಗಳಲ್ಲಿ ಒಂದಾಗಿದೆ. ಇವುಗಳಲ್ಲಿ ಇವುಗಳೂ ಸೇರಿವೆ: 

- ಮಲ್ಟಿಮೋಡಾಲಿಟಿ - ಪಠ್ಯ ಮತ್ತು ಚಿತ್ರ ಪ್ರಾಂಪ್ಟ್‌ಗಳ ಎರಡನ್ನೂ ಮೌಲ್ಯಮಾಪನ ಮಾಡಲು ಸಾಮರ್ಥ್ಯವಿದೆ 
- ಸಣ್ಣದಿಂದ ಮಧ್ಯಮ ಗಾತ್ರದ ಬದಲಾವಣೆಗಳು (11B ಮತ್ತು 90B) - ಇದು ಅನುಗುಣ ವಾಹನ ಆಯ್ಕೆಗಳನ್ನು ಒದಗಿಸುತ್ತದೆ, 
- ಪಠ್ಯ-ಮಾತ್ರ ಬದಲಾವಣೆಗಳು (1B ಮತ್ತು 3B) - ಇದು ಎಡ್ಜ್ / ಮೊಬೈಲ್ ಸಾಧನಗಳಲ್ಲಿ ಮಾದರಿಯನ್ನು ನಿಯೋಜಿಸಲು ಅನುಮತಿಸುತ್ತದೆ ಮತ್ತು ಕಡಿಮೆ ವಿಳಂಬವನ್ನು ಒದಗಿಸುತ್ತದೆ 

ಮಲ್ಟಿಮೋಡಲ್ ಬೆಂಬಲವು ಮುಕ್ತ ಮೂಲ ಮಾದರಿಗಳ ಜಗತ್ತಿನಲ್ಲಿ ದೊಡ್ಡ ಹೆಜ್ಜೆಯಾಗಿದೆ. ಕೆಳಗಿನ ಕೋಡ್ ಉದಾಹರಣೆ ಒಂದು ಚಿತ್ರ ಮತ್ತು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಎರಡನ್ನೂ ತೆಗೆದು ಲ್ಲಾಮಾ 3.2 90B ನಿಂದ ಚಿತ್ರದ ವಿಶ್ಲೇಷಣೆಯನ್ನು ಪಡೆಯುತ್ತದೆ. 

### ಲ್ಲಾಮಾ 3.2 ಯೊಂದಿಗೆ ಮಲ್ಟಿಮೋಡಲ್ ಬೆಂಬಲ


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
