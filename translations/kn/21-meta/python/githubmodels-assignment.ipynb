## ಲ್ಲಾಮಾ 3.1

405ಬಿಲಿಯನ್ ಪ್ಯಾರಾಮೀಟರ್ಗಳೊಂದಿಗೆ, ಬೇರೆಯ ಸ್ರೋತ LLM ವರ್ಗದಲ್ಲಿ ಲ್ಲಾಮಾ 3.1 ಸಿಗುತ್ತದೆ.

ಈ ಮೋಡ್ ಹಿಂದಿನ ಬಿಡುಗಡೆಯಾದ ಲ್ಲಾಮಾ 3 ಗೆ ಈ ಕೆಳಗಿನ ಸುಧಾರಣೆಗಳನ್ನು ನೀಡುತ್ತದೆ:

- ಹೆಚ್ಚಿನ ಪರಿಧಿ ಕಿಟಕಿ - 128ಕೆ ಟೋಕನ್ಸ್ ವಿರುದ್ಧ 8ಕೆ ಟೋಕನ್ಸ್
- ಹೆಚ್ಚಿನ ಗರಿಷ್ಠ ಔಟ್ಪುಟ್ ಟೋಕನ್ಸ್ - 4096 ವಿರುದ್ಧ 2048
- ಉತ್ತಮ ಬಾಹುಭಾಷಾ ಬೆಂಬಲ - ತರಬೇತಿ ಟೋಕನ್ಸ್ ಹೆಚ್ಚಳದಿಂದ

ಈವುಗಳು ಲ್ಲಾಮಾ 3.1 ಗೆ GenAI ಅಪ್ಲಿಕೇಶನ್‌ಗಳನ್ನು ನಿರ್ಮಿಸುವಾಗ ಹೆಚ್ಚು ಸಂಕೀರ್ಣ ಬಳಸುವ ಪ್ರಕರಣಗಳನ್ನು ನಿಭಾಯಿಸಲು ಸಾಧ್ಯವಾಗಿಸುತ್ತವೆ, ಈವುಗಳ ಒಳಗೊಂಡಂತೆ:
- ನೈಜ ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ - LLM ಕಾರ್ಯಪ್ರವಾಹದ ಹೊರಗಿನ ಬಾಹ್ಯ ಸಾಧನಗಳು ಮತ್ತು ಫಂಕ್ಷನ್‌ಗಳನ್ನು ಕಾಲ್ ಮಾಡುವ ಸಾಮರ್ಥ್ಯ
- ಉತ್ತಮ RAG ಕಾರ್ಯಕ್ಷಮತೆ - ಹೆಚ್ಚು ಪರಿಧಿ ಕಿಟಕಿಯಿಂದ
- ಸಿಂಥೆಟಿಕ್ ಡೇಟಾ ಉತ್ಪಾದನೆ - ಸೂಕ್ತವಾದ ಫೈನ್-ಟ್ಯೂನಿಂಗ್ ಕಾರ್ಯಗಳಿಗೆ ಡೇಟಾವನ್ನು ರಚಿಸುವ ಸಾಮರ್ಥ್ಯ



### ಸ್ವದೇಶಿ ಕಾರ್ಯಾಚರಣೆ ಕರೆ ಮಾಡುವುದು

ಲ್ಲಾಮಾ 3.1 ಅನ್ನು ಕಾರ್ಯ ಅಥವಾ ಉಪಕರಣ ಕರೆಗಳನ್ನು ಹೆಚ್ಚು ಪರಿಣಾಮಕಾರಿ ಮಾಡಲು ಸೂಕ್ಷ್ಮಗೊಳಿಸಲಾಗಿದೆ. ಇದು ಬಳಕೆದಾರರ ಪ್ರಾಂಪ್ಟ್‌ ಆಧಾರಿತವಾಗಿ ಬಳಸಬೇಕಾದಂತೆ ಗುರುತಿಸಬಹುದಾದ ಎರಡು ಒಳಗಿನ ಉಪಕರಣಗಳನ್ನು ಹೊಂದಿದೆ. ಈ ಉಪಕರಣಗಳು ಇವು: 

- **ಬ್ರೇವ್ ಸರ್ಚ್** - ವೆಬ್ ಸರ್ಚ್ ಮೂಲಕ ಹವಾಮಾನದ ನವೀಕೃತ ಮಾಹಿತಿಯನ್ನು ಪಡೆಯಲು ಬಳಸಬಹುದು 
- **ವೋಲ್ಫ್ರಾಂ ಅಲ್ಫಾ** - ಹೆಚ್ಚು ಸಂಕೀರ್ಣ ಗಣಿತೀಯ ಲೆಕ್ಕಾಚಾರಗಳಿಗೆ ಬಳಸಬಹುದು, ಆದಕಾರণে ನೀವು ನಿಮ್ಮದೇ ಕಾರ್ಯಗಳನ್ನು ಬರೆಯಬೇಕಿಲ್ಲ 

ನೀವು ಹೀಗಾಗಿ ನಿಮ್ಮದೇ ಕಸ್ಟಮ್ ಉಪಕರಣಗಳನ್ನು ರಚಿಸಬಹುದು, ಲ್ಲಾಮಾ ಆಲгоритಮ್ ಅವುಗಳನ್ನು ಕರೆ ಮಾಡಲಿದೆ. 

ಕೆಳಗಿನ ಕೊಡ್ ಉದಾಹರಣೆಯಲ್ಲಿ: 

- ನಾವು ಲಭ್ಯವಿರುವ ಉಪಕರಣಗಳನ್ನು (brave_search, wolfram_alpha) ಸಿಸ್ಟಮ್ ಪ್ರಾಂಪ್ಟ್‌ನಲ್ಲಿ ವ್ಯಾಖ್ಯಾನಿಸುತ್ತೇವೆ. 
- ನಿರ್ದಿಷ್ಟ ನಗರದ ಹವಾಮಾನದ ಬಗ್ಗೆ ಕೇಳುವ ಬಳಕೆದಾರ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಕಳುಹಿಸುತ್ತೇವೆ. 
- ಲ್ಲಾಮಾ ಬ್ರೇವ್ ಸರ್ಚ್ ಉಪಕರಣವನ್ನು ಕರೆಮಾಡುವಂತೆ ಪ್ರತಿಕ್ರಿಯಿಸುವುದು, ಉದಾಹರಣೆಗೆ `<|python_tag|>brave_search.call(query="Stockholm weather")` ಇದೆ 

*ಗಮನಿಸಿ: ಈ ಉದಾಹರಣೆ ಕೇವಲ ಉಪಕರಣವನ್ನು ಕರೆಮಾಡುತ್ತದೆ, ಫಲಿತಾಂಶಗಳನ್ನು ಪಡೆಯಲು, ನೀವು ಬ್ರೇವ್ API ಪುಟದಲ್ಲಿ ಉಚಿತ ಖಾತೆಯನ್ನು ರಚಿಸಿ ಕಾರ್ಯವನ್ನು ನಿರ್ವಹಿಸಬೇಕು* 


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### ಲ್ಲಾಮಾ 3.2 

ಎಲ್‌ಎಲ್‌ಎಂ ಆಗಿದ್ದರೂ ಸಹ, ಲ್ಲಾಮಾ 3.1ಗೆ ಇರುವ ಒಂದು ಮಿತಿ ಮಲ್ಟಿಮೋಡಾಲಿಟಿ ಆಗಿದೆ. ಅಂದರೆ, ಚಿತ್ರಗಳನ್ನು ಪ್ರಾಂಪ್ಟ್ ಆಗಿ ಬಳಸುವ ಮತ್ತು ಪ್ರತಿಕ್ರಿಯೆಗಳನ್ನು ನೀಡುವ ಹೋಲಿಕೆಯಿಂದ ವಿಭಿನ್ನ ರೀತಿಯ ಇನ್ಪುಟ್ಗಳನ್ನು ಬಳಸಲು ಸಾಧ್ಯತೆ. ಈ ಸಾಮರ್ಥ್ಯ ಲ್ಲಾಮಾ 3.2ನ ಪ್ರಮುಖ ವೈಶಿಷ್ಟ್ಯಗಳಲ್ಲೊಬ್ಬ. ಈ ವೈಶಿಷ್ಟ್ಯಗಳಲ್ಲಿ ಇವುಗಳೂ ಸೇರಿವೆ:

- ಮಲ್ಟಿಮೋಡಾಲಿಟಿ - ಪಠ್ಯ ಮತ್ತು ಚಿತ್ರ ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಎರಡನ್ನೂ ಮೌಲ್ಯಮಾಪನ ಮಾಡುವ ಸಾಮರ್ಥ್ಯ 
- ಸಣ್ಣದಿಂದ ಮಧ್ಯಮ ಗಾತ್ರದ ಬದಲಾವಣೆಗಳು (11B ಮತ್ತು 90B) - ಇದು ನಿಬಂಧಿತ ಉಪಯೋಗ ಆಯ್ಕೆಗಳು ಒದಗಿಸುತ್ತದೆ,
- ಪಠ್ಯ ಮಾತ್ರದ ಬದಲಾವಣೆಗಳು (1B ಮತ್ತು 3B) - ಇದರಿಂದ ಮಾದರಿಯನ್ನು ಎಡ್ಜ್ / ಮೊಬೈಲ್ ಸಾಧನಗಳಲ್ಲಿ ನಿಯೋಜಿಸಲು ಸಾಧ್ಯವಾಗುತ್ತದೆ ಮತ್ತು ಕಡಿಮೆ ವಿಳಂಬ ಒದಗಿಸುತ್ತದೆ 

ಮಲ್ಟಿಮೋಡಲ್ ಬೆಂಬಲವು ಮುಕ್ತ ಮೂಲ ಮಾದರಿಗಳ ಲೋಕದಲ್ಲಿ ದೊಡ್ಡ ಹೆಜ್ಜೆ ಎಂದು ಪ್ರತಿಭಾಸಿಸುತ್ತದೆ. ಕೆಳಗಿನ ಕೋಡ್ ಉದಾಹರಣೆ ಲ್ಲಾಮಾ 3.2 90B ನಿಂದ ಚಿತ್ರದ ವಿಶ್ಲೇಷಣೆ ಪಡೆಯಲು ಚಿತ್ರ ಮತ್ತು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಎರಡನ್ನೂ ತೆಗೆದುಕೊಳ್ಳುತ್ತದೆ.

### ಲ್ಲಾಮಾ 3.2 ಸಹಿತ ಮಲ್ಟಿಮೋಡಲ್ ಬೆಂಬಲ


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
