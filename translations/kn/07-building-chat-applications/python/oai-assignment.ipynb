#### ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಯನ್ನು ಹೇಗೆ ನಿಯಂತ್ರಿಸಬೇಕು  
"ಒಳಿದ ಎಲ್ಲಾ ಇನ್‌ಪುಟ್‌ಗಳಿಗಿಂತ, ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗೆ ಅತ್ಯಂತ ಪ್ರಭಾವಶೀಲವಾದುದು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್(1).

ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗಳನ್ನು ಕೆಲವು ರೀತಿಯಲ್ಲಿ ಪ್ರಾಂಪ್ಟ್ ಮಾಡಬಹುದು:

ಸೂಚನೆ: ನೀವು ಬೇಕಾದುದನ್ನು ಮಾದರಿಗೆ ಹೇಳಿ  
ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ: ನೀವು ಬೇಕಾದುದನ್ನು ಆರಂಭಿಕವಾಗಿ ಪೂರ್ಣಗೊಳಿಸಲು ಮಾದರಿಯನ್ನು ಪ್ರೇರೇಪಿಸಿ

#### ಪ್ರಾಂಪ್ಟ್ ರಚಿಸಲು ಮೂರು ಮೂಲಭೂತ ಮಾರ್ಗಸೂಚಿಗಳು ಇವೆ:

**ತೋರಿಸಿ ಮತ್ತು ಹೇಳಿ**. ಸೂಚನೆಗಳು, ಉದಾಹರಣೆಗಳು ಅಥವಾ ಇಬ್ಬರ ಸಂಯೋಜನೆಯ ಮೂಲಕ ನೀವು ಬೇಕಾದದ್ದಿನಲ್ಲಿ ಸ್ಪಷ್ಟವಾಗಿರಿ. ನೀವು ಮಾದರಿಯನ್ನು ಅಕ್ಷರ ಕ್ರಮದಲ್ಲಿ ಅಂಶಗಳ ಪಟ್ಟಿಯನ್ನು ಶ್ರೇಣಿಗೊಳಿಸಲು ಅಥವಾ ಪ್ಯಾರಾಗ್ರಾಫ್ ಅನ್ನು ಭಾವಗ್ರಹಣದ ಪ್ರಕಾರ ವರ್ಗೀಕರಿಸಲು ಬಯಸಿದರೆ, ಅದು ನೀವು ಬಯಸುವುದೇ ಅಂತ ತೋರಿಸಿ.  

**ಗುಣಮಟ್ಟದ ಡೇಟಾವನ್ನು ಒದಗಿಸಿ**. ನೀವು ವರ್ಗೀಕರಣೆ ಯಂತ್ರ ಅಥವಾ ಮಾದರಿಯು ಒಂದು ಮಾದರಿಯನ್ನು ಅನುಸರಿಸಲು ಪ್ರಯತ್ನಿಸುತ್ತಿದ್ದರೆ, ಸಾಕಷ್ಟು ಉದಾಹರಣೆಗಳಿರಬೇಕು. ನಿಮ್ಮ ಉದಾಹರಣೆಗಳನ್ನು ಪರಿಶೀಲಿಸಿ — ಸಾಮಾನ್ಯವಾಗಿ ಮಾದರಿ ಮೂಲಭೂತ ವ್ಯಾಕರಣ ದೋಷಗಳನ್ನು ಗಮನಿಸಲಿಕ್ಕೆ ಸಾಕಷ್ಟು ಬುದ್ಧಿವಂತಿಕೆಯುಳ್ಳದ್ದು ಮತ್ತು ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ನೀಡುತ್ತದೆ, ಆದರೆ ಇದು ತಿಳಿದುಕೊಂಡು ವೈಯಕ್ತಿಕ ಬುದ್ಧಿಪೂರ್ವಕ ಎಂದು ಭಾವಿಸಬಹುದು ಮತ್ತು ಪ್ರತಿಕ್ರಿಯೆಗೆ ಪ್ರಭಾವ ಬೀರುತ್ತದೆ.  

**ನಿಮ್ಮ ಸೆಟ್ಟಿಂಗ್‌ಗಳನ್ನು ಪರಿಶೀಲಿಸಿ**. ತಂಪಾದೆ ಮತ್ತು top_p ಸೆಟ್ಟಿಂಗ್‌ಗಳು ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಉತ್ಪಾದಿಸುವಲ್ಲಿ ಮಾದರಿಯ ನಿಶ್ಚಿತತೆಯನ್ನು ನಿಯಂತ್ರಿಸುತ್ತವೆ. ನೀವು ಒಬ್ಬ ಶುದ್ಧ ಉತ್ತರಕ್ಕಾಗಿ ಕೇಳುತ್ತಿದ್ದರೆ, ಅವುಗಳನ್ನು ಕೆಳಗೆ ಸೆಟ್ ಮಾಡಬೇಕು. ನೀವು ಹೆಚ್ಚು ವೈವಿಧ್ಯಮಯ ಪ್ರತಿಕ್ರಿಯೆಗಳಿಗಾಗಿ ಹುಡುಕುತ್ತಿದ್ದರೆ, ಅವುಗಳನ್ನು ಮೇಲಕ್ಕೆ ಸೆಟ್ ಮಾಡಬಹದು. ಜನರು ಈ ಸೆಟ್ಟಿಂಗ್‌ಗಳನ್ನು “ಬುದ್ಧಿಮತ್ತೆ” ಅಥವಾ “ಕಲ್ಪನೆ” ನಿಯಂತ್ರಣ ಎಂದು ಭ್ರಮಿಸುವುದು ದೋಷವೆಂದೇ ವಿಷಯವಾಗಿದೆ.
