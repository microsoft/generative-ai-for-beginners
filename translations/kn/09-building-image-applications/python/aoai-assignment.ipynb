# ಚಿತ್ರ ನಿರ್ಮಾಣ ಅಪ್ಲಿಕೇಶನ್ಗಳನ್ನು ನಿರ್ಮಿಸುವುದು (Azure OpenAI)

LLM ಗಳಿಗೆ ಬರೆಹ ತಯಾರಿಕೆಯ ಮೇಲೆ ಹೆಚ್ಚುವರಿ ಬಲವಿದೆ. ನೀವು ಬರೆಹ ವಿವರಣೆಗಳಿಂದ ತೋರ್ಪಡಿಸುವ ಚಿತ್ರಗಳನ್ನು ಕೂಡ ನಿರ್ಮಿಸಬಹುದು. ಚಿತ್ರಗಳು ಒಂದು ಮಾಧ್ಯಮವಾಗಿ ಮೆಡ್ ಟೆಕ್, ವಾಸ್ತುಶಿಲ್ಪ, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟದ ಅಭಿವೃದ್ಧಿ, ಮಾರುಕಟ್ಟೆ ಮತ್ತು ಇತರೆ ಕ್ಷೇತ್ರಗಳಲ್ಲಿ ಉಪಯುಕ್ತವಾಗಿವೆ. ಈ ಪಾಠದಲ್ಲಿ ನಾವು Microsoft Foundry ನಲ್ಲಿ ಇಂದಿನ **GPT Image** ಮಾದರಿಗಳನ್ನು ಬಳಸುತ್ತೇವೆ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠದ ಕೊನೆಯಲ್ಲಿ ನೀವು ಸಾಧ್ಯವಾಗುವುದು:

- ಚಿತ್ರ ತಯಾರಿಕೆ ಎಂದರೇನು ಮತ್ತು ಅದು ಎಲ್ಲಿಗೆ ಉಪಯುಕ್ತವೋ ವಿವರಿಸುವುದು.
- `gpt-image` ಮಾದರಿ ಕುಟುಂಬವನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು ಮತ್ತು ಅದು ಹಳೆಯ DALL·E ಮಾದರಿಗಳಿಂದ ಹೇಗೆ ಭಿನ್ನವಾಗಿದ್ದುದು ಎಂಬುದು.
- ಚಿತ್ರ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ, ಚಿತ್ರಗಳನ್ನು ಸಂಪಾದಿಸುವುದು.

## ಚಿತ್ರ ತಯಾರಿಕೆ ಎಂದರೇನು?

ಚಿತ್ರ ತಯಾರಿಕೆ ಮಾದರಿಗಳು ಬರೆಹ ಪ್ರಾಂಪ್ಟ್ ನಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸುತ್ತವೆ. ಆಧುನಿಕ ಮಾದರಿಗಳು, ಉದಾಹರಣೆಗೆ `gpt-image`, ತರಬೇತಿ ಸಮಯದಲ್ಲಿ ಬರೆಹ ಮತ್ತು ಚಿತ್ರಗಳ ನಡುವಿನ ಸಂಬಂಧವನ್ನು ಕಲಿಯುತ್ತವೆ, ನಂತರ ಕ್ರಮೇಣ ಯಾದೃಚ್ಛಿಕ ಶಬ್ದವನ್ನು ಚಿತ್ರವಾಗಿ ರೂಪಾಂತರಿಸುತ್ತವೆ ಇದು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ಗೆ ಹೊಂದಿಕೊಳ್ಳುತ್ತದೆ.

ಎರಡು ಜನಪ್ರಿಯ ಚಿತ್ರ ಮಾದರಿ ಕುಟುಂಬಗಳು:

- **`gpt-image` (OpenAI)** — ಈ ಪಾಠದಲ್ಲಿ ಬಳಸಲಾಗುವ ಪ್ರಸ್ತುತ ತಲೆಮಾರಿನ ಮಾದರಿ. ಇದು ಬರೆಹದಿಂದ ಚಿತ್ರಕ್ಕೆ ತಯಾರಿಕೆ ಮತ್ತು ಚಿತ್ರ ಸಂಪಾದನೆಯ (ಮಾಸ್ಕ್‌ನಿಂದ ಇನ್‌ಪೇಂಟಿಂಗ್) ಕಾರ್ಯಗಳನ್ನು ಬೆಂಬಲಿಸುತ್ತದೆ.
- **Midjourney** — ಸ್ವಂತ ಸೇವೆ ಮತ್ತು Discord ಆಧಾರಿತ ಕೆಲಸದ ಸಾಧ್ಯತೆ ಇರುವ ಜನಪ್ರಿಯ ತೃತೀಯ ಪಕ್ಷದ ಮಾದರಿ.

> ಹಳೆಯ OpenAI ಚಿತ್ರ ಮಾದರಿ — **DALL·E 2** ಮತ್ತು **DALL·E 3** — ಹಳೆಯವುಗಳಾಗಿವೆ. DALL·E 3 ನವೀಕೃತ ನಿಯೋಜನೆಗಳಿಗೆ ಲಭ್ಯವಿಲ್ಲ, ಮತ್ತು `create_variation` ಹಂತಹಂತವು ಕೇವಲ DALL·E 2 ನಲ್ಲಿ ಮಾತ್ರ ಇತ್ತು. ಹೊಸ ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ `gpt-image` ಮಾದರಿಗಳನ್ನು ಬಳಸಿ.

Microsoft Foundry ನಲ್ಲಿ, **`gpt-image-2`** ತಾಜಾ ಮತ್ತು ಅತ್ಯಂತ ಸಾಮರ್ಥ್ಯವಂತಿಕೆ ಇರುವ ಚಿತ್ರ ಮಾದರಿಯಾಗಿದೆ ಮತ್ತು ಶಿಫಾರಸು ಮಾಡಿದ ಮೂಲ ಮಾದರಿ. `gpt-image-1.5` ಮತ್ತು `gpt-image-1-mini` ಕೂಡ ಸಾಮಾನ್ಯವಾಗಿ ಲಭ್ಯವಿವೆ.

> **ಮುಖ್ಯವಿಷಯ:** `gpt-image` ಮಾದರಿಗಳು ರಚಿಸಿದ ಚಿತ್ರವನ್ನು **base64** (`b64_json`) ರೂಪದಲ್ಲಿ ನೀಡುತ್ತವೆ, URL ರೂಪದಲ್ಲಿ ಅಲ್ಲ. ನಿಮ್ಮ ಕೋಡ್ base64 ಸರಣಿಯನ್ನು ಡಿಕೋಡ್ ಮಾಡಿ ಬೈಟ್‌ಗಳಾಗಿ ಉಳಿಸುತ್ತದೆ — ಡೌನ್‌ಲೋಡ್ ಮಾಡಲು ಯಾವುದೇ ಚಿತ್ರ URL ಇಲ್ಲ.


## ನಿಮ್ಮ ಮೊದಲ ಚಿತ್ರ ಉತ್ಪಾದನಾ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ನಿರ್ಮಿಸುವುದು

ಐಮೇಜ್ ಜನರೇಶನ್ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ನಿರ್ಮಿಸಲು ಏನು ಬೇಕಾಗುತ್ತದೆ? ನಿಮಗೆ ಕೆಳಗಿನ ಲೈಬ್ರರಿಗಳು ಬೇಕಾಗಿವೆ:

- **python-dotenv**, ನಿಮ್ಮ ರಹಸ್ಯಗಳನ್ನು ಕೋಡ್‌ನಿಂದ ದೂರದಲ್ಲಿರುವ *.env* ಫೈಲಿನಲ್ಲಿ ಇರಿಸಲು ಈ ಲೈಬ್ರರಿ ಬಳಸಲು ಶಿಫಾರಸು ಮಾಡಲಾಗಿದೆ.
- **openai**, ನೀವು OpenAI API ಜೊತೆ ಸಂವಹನ ಮಾಡಲು ಬಳಸುವ ಲೈಬ್ರರಿ.
- **pillow**, Python ನಲ್ಲಿ ಚಿತ್ರಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು.

ಈಗಾಗಲೇ ಮಾಡದಿದ್ದರೆ, Microsoft Foundry ಸಂಪನ್ಮೂಲ ಮತ್ತು ಮಾದರಿಯನ್ನು ರಚಿಸಲು [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ಪುಟದಲ್ಲಿ ಸೂಚನೆಗಳನ್ನು ಅನುಸರಿಸಿ. ಮಾದರಿಯಾಗಿ **gpt-image-2** ಅನ್ನು ಆರಿಸಿಕೊಳ್ಳಿ (ಇದು ಇತ್ತೀಚಿನ Azure OpenAI ಚಿತ್ರ ಮಾದರಿ; DALL·E ಹಳೆಯ ಮಾದರಿ).

1. ಕೆಳಗಿನ ವಿಷಯದೊಂದಿಗೆ *.env* ಫೈಲ್ ರಚಿಸಿ:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    ಈ ಮಾಹಿತಿಯನ್ನು ನಿಮ್ಮ ಸಂಪನ್ಮೂಲದ "Deployments" ವಿಭಾಗದಲ್ಲಿ Microsoft Foundry ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ಕಂಡುಹಿಡಿಯಿರಿ.



1. ಮೇಲಿನ ಲೈಬ್ರರಿಗಳನ್ನು *requirements.txt* ಎಂದು ಹೆಸರಿಸಿರುವ ಫೈಲ್‌ನಲ್ಲಿ ಈ ರೀತಿ ಸಂಗ್ರಹಿಸಿ:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. ಮುಂದಿನ ಹಂತವಾಗಿ, ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಿ ಮತ್ತು ಲೈಬ್ರರಿಗಳನ್ನು ಸ್ಥಾಪಿಸಿ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> ವಿಂಡೋಸ್‌ಗಾಗಿ, ನಿಮ್ಮ ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಲು ಮತ್ತು ಸಕ್ರಿಯಗೊಳಿಸಲು ಕೆಳಗಿನ ಆಜ್ಞೆಗಳನ್ನು ಬಳಸಿ:


1. *app.py* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಕೆಳಗಿನ ಕೋಡ್ ಅನ್ನು ಸೇರಿಸಿ:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # ಡಾಟ್‌ಎನ್‌ವಿ आयಾಂಪ್
    dotenv.load_dotenv()

    # ಅಝುರ್ ಓಪನ್‌ಎಐ (ಮೈಕ್ರೋಸಾಫ್ಟ್ ಫೌಂಡ್ರಿ) ಕ್ಲೈಯಂಟನ್ನು ಕಾಂಫಿಗರ್ ಮಾಡಿ.
    # ಚಿತ್ರ ಮಾದರಿಗಳಿಗೆ ಇತ್ತೀಚಿನ API ಆವೃತ್ತಿ ಬೇಕಾಗಿರುತ್ತದೆ — ನಿಮ್ಮ ಮಾದರಿ ಅಗತ್ಯವಿರುವುದನ್ನು ಫೌಂಡ್ರಿ ಡಾಕ್ಸ್‌ನಲ್ಲಿ ಪರಿಶೀಲಿಸಿ.
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # ಚಿತ್ರ ರಚಿಸಲು ಚಿತ್ರ ಸೃಷ್ಟಿ API ಬಳಸಿ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # ಉದಾಹರಣೆಗೆ: "gpt-image-2"
        )
        # ಸಂಗ್ರಹಿಸಲಾದ ಚಿತ್ರದ ಡೈರೆಕ್ಟರಿಯನ್ನು ಸೆಟ್ ಮಾಡಿ
        image_dir = os.path.join(os.curdir, 'images')

        # ಡೈರೆಕ್ಟರಿಯು ಇಲ್ಲದಿದ್ದರೆ, ಅದನ್ನು ರಚಿಸಿ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ಚಿತ್ರದ ಪಥವನ್ನು ಪ್ರಾರಂಭಿಸಿ (ಫೈಲ್ ಪ್ರಕಾರ PNG ಆಗಿರಬೇಕು ಎಂದು ಗಮನಿಸಿ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image ಮಾದರಿಗಳು URL ಅಲ್ಲದೆ ಚಿತ್ರವನ್ನು base64 (b64_json) ಆಗಿ ನೀಡುತ್ತವೆ
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ಡೀಫಾಲ್ಟ್ ಚಿತ್ರ ವೀಕ್ಷಕದಲ್ಲಿ ಚಿತ್ರದ ಪ್ರದರ್ಶನ ಮಾಡಿ
        image = Image.open(image_path)
        image.show()

    # ಅಪವಾದಗಳನ್ನು ಸೆರೆಹಿಡಿಯಿರಿ
    except BadRequestError as err:
        print(err)

    ```

ಈ ಕೋಡ್ ಅನ್ನು ವಿವರಿಸೋಣ:

- ಮೊದಲು, ನಮಗೆ ಬೇಕಾದ ಗ್ರಂಥಾಲಯಗಳನ್ನು ಆಮದು ಮಾಡಿಕೊಳ್ಳುತ್ತೇವೆ, ಇದರಲ್ಲಿ OpenAI ಗ್ರಂಥಾಲಯ, dotenv ಗ್ರಂಥಾಲಯ, base64 ಮ್ಯಾಡ್ಯೂಲ್ ಮತ್ತು Pillow ಗ್ರಂಥಾಲಯ ಸೇರಿವೆ.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- ನಂತರ, *.env* ಫೈಲ್‌ನಿಂದ ಪರಿಸರ ಚರಗಳನ್ನು ಲೋಡ್ ಮಾಡುತ್ತೇವೆ.

    ```python
    # ಡಾಟ್‌ಎನ್‌ವಿ ಅನ್ನು ಆಮದುಮಾಡಿ
    dotenv.load_dotenv()
    ```

- ಆ ನಂತರ, Azure OpenAI (Microsoft Foundry) ಕ್ಲೈಯಂಟ್ ಅನ್ನು ಸಂರಚಿಸುತ್ತೇವೆ.

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- ನಂತರ, ಚಿತ್ರವನ್ನು ಉತ್ಪಾದಿಸುತ್ತೇವೆ:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` ಮಾದರಿಗಳು ಚಿತ್ರವನ್ನು **base64** ಸ್ಟ್ರಿಂಗ್ ಆಗಿ `data[0].b64_json` ನಲ್ಲಿ ನೀಡುತ್ತವೆ. ನಾವು ಅದನ್ನು ಬೈಟ್ಸ್‌ಗೆ ಡಿಕೋಡ್ ಮಾಡಿ ಫೈಲ್‌ಗೆ ಬರೆಯುತ್ತೇವೆ — ერಡುತಿಗಾಗಿ ಯಾವುದೇ URL ಇರುವುದಿಲ್ಲ.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- ಕೊನೆಗೆ, ಚಿತ್ರವನ್ನು ತೆರೆಯುತ್ತೇವೆ ಮತ್ತು ಸಾಮಾನ್ಯ ಚಿತ್ರ ವೀಕ್ಷಕವನ್ನು ಬಳಸಿ ಅದನ್ನು ತೋರಿಸುತ್ತೇವೆ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ಚಿತ್ರವನ್ನು ಉತ್ಪಾದಿಸುವ ಬಗ್ಗೆ ಇನ್ನಷ್ಟು ವಿವರಗಳು

`images.generate` ಪಾರಾಮೀಟರ್‌ಗಳನ್ನು ನೋಡೋಣ:

- **prompt**, ಚಿತ್ರವನ್ನು ಉತ್ಪಾದಿಸಲು ಬಳಸುವ ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್. ಇಲ್ಲಿ ಅದು "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, ಉತ್ಪಾದಿತ ಚಿತ್ರದ ಗಾತ್ರ (ಉದಾಹರಣೆಗೆ `1024x1024`, `1536x1024`, `1024x1536`, ಅಥವಾ `"auto"`).
- **n**, ಉತ್ಪಾದಿಸಲಾದ ಚಿತ್ರಗಳ ಸಂಖ್ಯೆಯನ್ನು ಸೂಚಿಸುತ್ತದೆ. ಇಲ್ಲಿ ನಾವು ಒಂದು ಚಿತ್ರವನ್ನು ಉತ್ಪಾದಿಸುತ್ತೇವೆ.
- **model**, ನಿಮ್ಮ ಚಿತ್ರ ಮಾದರಿ ಪ್ರತ್ಯೇಕಿಸುವ ಹೆಸರಾಗಿದೆ (ಉದಾಹರಣೆಗೆ `gpt-image-2`).

> ಚಿತ್ರ ಮಾದರಿಗಳು `temperature` ಪ್ಯಾರಾಮೀಟರ್ ತೆಗೆದುಕೊಳ್ಳುವುದಿಲ್ಲ — ಅದು ಪಠ್ಯ-ಉತ್ಪಾದನೆ ನಿಯಂತ್ರಣವಾಗಿದೆ. ವಿಭಿನ್ನತೆ ಕೈಗೊಳ್ಳಲು, API ಅನ್ನು ಮತ್ತೆ ಕರೆಯಿರಿ; ವಿಭಿನ್ನತೆಯನ್ನು ಕಡಿಮೆ ಮಾಡಲು, ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಹೆಚ್ಚು ನಿರ್ದಿಷ್ಟವಾಗಿಸಿ.

## ಚಿತ್ರ ಉತ್ಪಾದನೆಯ ಹೆಚ್ಚುವರಿ ಸಾಮರ್ಥ್ಯಗಳು

ನೀವು ಕೆಲವು ಸಾಲುಗಳ ಪೈಥಾನ್ ಮೂಲಕ ಚಿತ್ರ ಅನ್ನು ಹೇಗೆ ಉತ್ಪಾದಿಸಬಹುದು ಎಂಬುದನ್ನು ನೋಡಿದ್ದೀರಿ. `gpt-image` ಮಾದರಿಗಳು ಇದ್ದ ಚಿತ್ರವನ್ನು **ತಿದ್ದು** ಮಾಡಬಹುದು. ಒಂದು ಚಿತ್ರವನ್ನು ನೀಡುವ ಮೂಲಕ, ಐಚ್ಛಿಕ **ಮಾಸ್ಕ್** (ಬದಲಾಯಿಸಬೇಕಾದ ಪ್ರದೇಶ ಗುರುತಿಸುವುದು) ಮತ್ತು ಪ್ರಾಂಪ್ಟ್ ಒದಗಿಸಿದರೆ, ನೀವು ಚಿತ್ರದ ಭಾಗವನ್ನು ಬದಲಾಯಿಸಬಹುದು — ಉದಾಹರಣೆಗೆ, ನಮ್ಮ ಕೂರಿಗೆಯನ್ನು ಹೊಡೆಯಲು ಟೊಪಿಯನ್ನು ಸೇರಿಸಬಹುದು.

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ಸಂಪಾದನೆಗಳು ಕೂಡಾ ಬೇಸ್64 ಆಗಿ ಹಿಂತಿರುಗಿಸಲ್ಪಡುತ್ತವೆ
edited_image = base64.b64decode(response.data[0].b64_json)
```

ಮೂಲ ಚಿತ್ರದಲ್ಲಿ ಕೇವಲ ಮುಳ್ಳುಕುಳಿತಿದ್ದು; ಅಂತಿಮ ಚಿತ್ರದಲ್ಲಿ ಮುಳ್ಳುಕುಳಿತ ಮೇಲೆ ಟೊಪಿಯಿದೆ.
