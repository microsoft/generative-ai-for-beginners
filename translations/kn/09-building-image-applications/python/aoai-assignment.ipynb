# ಚಿತ್ರ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲಾಗುತ್ತಿದೆ 

LLM ಗಳು ಪಠ್ಯ ಉತ್ಪಾದನೆಗೆ ಮಾತ್ರವಲ್ಲದೆ ಇನ್ನಷ್ಟು ಕಾರ್ಯಗಳಿವೆ. ಪಠ್ಯ ವಿವರಣೆಗಳ ಮೂಲಕ ಚಿತ್ರಗಳನ್ನು ನಿರ್ಮಿಸುವುದೂ ಸಾಧ್ಯ. ಚಿತ್ರಗಳನ್ನು ಮಾಧ್ಯಮವಾಗಿ ಹೊಂದಿರುವುದು ಮೆಡ್ ಟೆಕ್, ವಾಸ್ತುಶಿಲ್ಪ, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಇತ್ಯಾದಿ ಹಲವಾರು ಕ್ಷೇತ್ರಗಳಲ್ಲಿ ಅತ್ಯಂತ ಉಪಯುಕ್ತವಾಗಬಹುದು. ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನಾವು ಅತ್ಯಂತ ಜನಪ್ರಿಯ ಚಿತ್ರ ತಯಾರಿಕೆ ಮಾದರಿಗಳಾದ DALL-E ಮತ್ತು Midjourney ಕುರಿತು ಪರಿಶೀಲನೆ ಮಾಡೋಣ.

## ಪರಿಚಯ 

ಈ ಪಾಠದಲ್ಲಿ, ನಾವು ನೋಡೋಣ:

- ಚಿತ್ರ ತಯಾರಿಕೆ ಮತ್ತು ಅದರ ಉಪಯುಕ್ತತೆ.
- DALL-E ಮತ್ತು Midjourney ಎಂದರೇನು, ಅವು ಹೇಗೆ ಕೆಲಸ ಮಾಡುತ್ತವೆ.
- ನೀವು ಚಿತ್ರ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸುವಿರಿ.

## ಅಧ್ಯಯನದ ಗುರಿಗಳು 

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ನಂತರ, ನೀವು ಸಾಧ್ಯವಾಗುತ್ತದೆ:

- ಚಿತ್ರ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ನಿರ್ಮಿಸು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಗಾಗಿ ಮೇಟಾ ಪ್ರಾಂಪ್ಟ್ ಗಳ ಮೂಲಕ ಗಡಿಗಳನ್ನು ನಿರ್ಧರಿಸು.
- DALL-E ಮತ್ತು Midjourney ನಲ್ಲಿ ಕೆಲಸ ಮಾಡು.

## ಚಿತ್ರ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಏಕೆ ನಿರ್ಮಿಸಬೇಕು?

ಚಿತ್ರ ತಯಾರಿಕಾ ಅಪ್ಲಿಕೇಶನ್ ಗಳು ಜನರೇಟಿವ್ AI ಶಕ್ತಿ ಅನ್ಸಲು ಒಳ್ಳೆಯ ಮಾರ್ಗವಾಗಿವೆ. ಉದಾಹರಣೆಗೆ ಇವು ಬಳಸಬಹುದು:  

- **ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಸಂಯೋಜನೆ**. ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಚಿತ್ರ ಸಂಯೋಜನೆ ಸೇರಿದಂತೆ ವಿವಿಧ ಬಳಕೆ ಪ್ರಕರಣಗಳಿಗಾಗಿ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸಬಹುದು.  

- **ವಿವಿಧ ಕೈಗಾರಿಕೆಗಳಿಗೆ ಅನ್ವಯಿಸುವುದಾಗಿದೆ**. ಮೆಡ್‌ಟೆಕ್, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಮತ್ತು ಇನ್ನಷ್ಟು ಕೈಗಾರಿಕೆಗಳಿಗೆ ಚಿತ್ರಗಳನ್ನು ತಯಾರಿಸಲು ಬಳಸಬಹುದು.

## DALL-E ಮತ್ತು Midjourney ಎಂದರೇನು? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) ಮತ್ತು [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) ಜತೆಗೆ ಅತ್ಯಂತ ಜನಪ್ರಿಯ ಚಿತ್ರ ತಯಾರಿಕೆಯ ಮಾದರಿಗಳು, ಇವು ಪ್ರಾಂಪ್ಟ್ ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸಲು ಸಾಧ್ಯವಿದೆ.

### DALL-E

DALL-E ಇಂದ ಪ್ರಾರಂಭಿಸೋಣ, ಇದು ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸುವ ಜನರೇಟಿವ್ AI ಮಾದರಿ. 

> [DALL-E ಎರಡು ಮಾದರಿಗಳ ಸಂಯೋಜನೆ, CLIP ಮತ್ತು ಡಿಫ್ಯೂಸ್_attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **CLIP**, ಚಿತ್ರಗಳು ಮತ್ತು ಪಠ್ಯದಿಂದ ಡೇಟಾಗೆ ಸಂಖ್ಯಾತ್ಮಕ ರೂಪಾಂತರಗಳನ್ನು ತಯಾರಿಸುವ ಇಂಬೆಡಿಂಗ್ಸ್ ಮಾಡಿದ ಮಾದರಿ.  

- **ಡಿಫ್ಯೂಸ್_attention**, ಇಂಬೆಡಿಂಗ್ಸ್‌ನಿಂದ ಚಿತ್ರಗಳನ್ನು ತಯಾರಿಸುವ ಮಾದರಿ. DALL-E ಚಿತ್ರಗಳು ಮತ್ತು ಪಠ್ಯದ ಡೇಟಾಸೆಟ್ ಮೇಲೆ ತರಬೇತುಗೊಂಡಿದ್ದು, ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸಲು ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ, DALL-E ಗೂದಲಿಟ್ಟು ಶೈಲಿಯ ಬ್ರಿಮ್ ಕ್ಯಾರ್ನಿವಲ್ ಉಡುಪು ಹಾಕಿದ ಬೆಕ್ಕಿನ ಚಿತ್ರ ಅಥವಾ ಮುಕ್ಕುಚಿತ್ತ ನಾಯಿ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸಬಹುದು. 

### Midjourney
 
Midjourney ಕೂಡ DALL-E ಯಂತೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ, ಇದು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸುತ್ತದೆ. Midjourney, “ಬ್ರಿಮ್ ಹ್ಯಾಟ್ ಹಾಕಿರುವ ಬೆಕ್ಕು” ಅಥವಾ “ಮುಕ್ಕುಚಿತ್ತ ನಾಯಿ” ಇಂತಹ ಪ್ರಾಂಪ್ಟ್ ಗಳನ್ನು ಬಳಸಿ ಚಿತ್ರಗಳನ್ನು ಸೃಷ್ಟಿಸಲು ಸಹ ಸಾಧ್ಯ.

## DALL-E ಮತ್ತು Midjourney ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತವೆ 

ಮೊದಲು, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ಒಂದು ಜನರೇಟಿವ್ AI ಮಾದರಿ ಇದು ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ಆರ್ಕಿಟೆಕ್ಟರ್ ಮೇಲೆ ಆಧಾರಿತವಾಗಿದ್ದು *ಆಟೋರಿಗ್ರೆಸಿವ್ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್* ಹೊಂದಿದೆ.  

ಒಂದು *ಆಟೋರಿಗ್ರೆಸಿವ್ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್* ಒಂದು ಮಾದರಿಯು ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರವನ್ನು ಹೇಗೆ ಉತ್ಪಾದಿಸುತ್ತದೆ ಎಂದು ನಿರ್ಧರಿಸುತ್ತದೆ, ಇದು ಒಂದು ಪಿಕ್ಸೆಲ್ನ್ನು ಸಮಯಕ್ಕೆ ನಿರ್ಮಿಸಿ ನಂತರ ಸೃಷ್ಟಿಸಿದ ಪಿಕ್ಸೆಲ್ಗಳನ್ನು ಬಳಸಿ ಮುಂದಿನ ಪಿಕ್ಸೆಲ್ನ್ನು ತಯಾರಿಸುತ್ತದೆ. ನ್ಯೂರಲ್ ನೆಟ್‌ವರ್ಕ್ ನಲ್ಲಿ ಹಲವು ತಹಗಳ ಮೂಲಕ ಹೋಗುತ್ತದೆ, ನಂತರ ಚಿತ್ರ ಪೂರ್ಣಗೊಳ್ಳುತ್ತದೆ.  

ಈ ಪ್ರಕ್ರಿಯೆಯಿಂದ, DALL-E ಚಿತ್ರದಲ್ಲಿ ಗುಣಲಕ್ಷಣಗಳು, ವಸ್ತುಗಳು, ಲಕ್ಷಣಗಳು ಇತ್ಯಾದಿಗಳನ್ನು ನಿಯಂತ್ರಿಸುತ್ತದೆ. ಆದಾಗ್ಯೂ, DALL-E 2 ಮತ್ತು 3 ಚಿತ್ರ ನಿರ್ಮಾಣದ ಮೇಲೆ ಹೆಚ್ಚು ನಿಯಂತ್ರಣ ಹೊಂದಿವೆ, 


## ನಿಮ್ಮ ಮೊದಲ ಚಿತ್ರ ಸೃಜನ ಅನ್ವಯಿಕೆಯನ್ನು ನಿರ್ಮಿಸುವುದು

ಆದಾಗ್ಯೂ ಚಿತ್ರ ಸೃಜನ ಅನ್ವಯಿಕೆಯನ್ನು ನಿರ್ಮಿಸಲು ಏನು ಬೇಕಾಗುತ್ತದೆ? ನಿಮಗೆ ಕೆಳಗಿನ ಗ್ರಂಥಾಲಯಗಳು ಬೇಕಾಗಿವೆ:

- **python-dotenv**, ನಿಮ್ಮ ರಹಸ್ಯಗಳನ್ನು ಕೋಡ್‌ನಿಂದ ದೂರವಿರುವ *.env* ಕಡತದಲ್ಲಿ ಇಡಲು ಈ ಗ್ರಂಥಾಲಯವನ್ನು ಬಳಸಲು ನೀವು ಬಹುಶಃ ಶಿಫಾರಸು ಮಾಡಲಾಗಿದೆ.
- **openai**, ನೀವು OpenAI API ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಲು ಈ ಗ್ರಂಥಾಲಯವನ್ನು ಬಳಸುವಿರಿ.
- **pillow**, Python ನಲ್ಲಿ ಚಿತ್ರಗಳೊಂದಿಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸಲು.
- **requests**, HTTP ವಿನಂತಿಗಳನ್ನು ಮಾಡಲು ಸಹಾಯ ಮಾಡಲು.

ಇನ್ನುವರೆಗೆ ಮಾಡದಿದ್ದರೆ, [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ಪುಟದಲ್ಲಿ ನೀಡಲಾದ ಸೂಚನೆಗಳನ್ನು ಅನುಸರಿಸಿ Azure OpenAI ಸಂಪನ್ಮೂಲ ಮತ್ತು ಮಾದರಿಯನ್ನು ರಚಿಸಿ. ಮಾದರಿಯಾಗಿ **gpt-image-1** ಆಯ್ಕೆಮಾಡಿ (ಪ್ರಸ್ತುತ ತಲೆಮಾರು Azure OpenAI ಚಿತ್ರ ಮಾದರಿ; DALL-E 3 ಹಳೆಯದು ಮತ್ತು ಹೊಸ ನಿಯೋಜನೆಗಳಿಗೆ ಲಭ್ಯವಿಲ್ಲ).

1. ಕೆಳಗಿನ ವಿಷಯವಿರುವ *.env* ಎಂಬ ಕಡತವನ್ನು ರಚಿಸಿ:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    ಈ ಮಾಹಿತಿಯನ್ನು ನಿಮ್ಮ ಸಂಪನ್ಮೂಲದ "Deployments" ವಿಭಾಗದಲ್ಲಿ Microsoft Foundry ಪೋರ್ಟಲ್‌ನಲ್ಲಿ ಕಂಡುಹಿಡಿಯಿರಿ.



1. ಮೇಲಿನ ಲೈಸರರಿಗಳನ್ನು *requirements.txt* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಕೆಳಗಿನಂತೆ ಸಂಗ್ರಹಿಸಿ:

    ```text

    requests
    ```


1. ಮುಂದಿನ ಹಂತದಲ್ಲಿ, ವರ್ಚುಯಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಿ ಮತ್ತು ಲೈಸರರಿಗಳನ್ನು ಸ್ಥಾಪಿಸಿ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````


1. *app.py* ಎಂದು ಕರೆಸುವ ಫೈಲ್‌ನಲ್ಲಿ ಕೆಳಗಿನ ಕೋಡ್ ಅನ್ನು ಸೇರಿಸಿ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv ಅನ್ನು ಆಮದುಮಾಡಿ
    dotenv.load_dotenv()

    # Azure OpenAI ಸೇವಾ ಕ್ಲಯಂಟ್ ಅನ್ನು ಸಂರಚಿಸಿ
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # ಚಿತ್ರ ರಚನೆ API ಬಳಸಿ ಚಿತ್ರವನ್ನು ಸೃಷ್ಟಿಸಿ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮूदಿಸಿ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # ಸಂಗ್ರಹಿಸಲಾದ ಚಿತ್ರದ ಡೈರೆಕ್ಟರಿಯನ್ನು ಹೊಂದಿಸಿ
        image_dir = os.path.join(os.curdir, 'images')
    
        # ಡೈರೆಕ್ಟರಿ ಇಲ್ಲದಿದ್ದರೆ ಅದನ್ನು ರಚಿಸಿ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # ಚಿತ್ರ ಮಾರ್ಗವನ್ನು ಪ್ರಾರಂಭಿಸಿ (ಫೈಲ್ ಟೈಪ್ png ಆಗಿರಬೇಕು ಎಂಬುದನ್ನು ಗಮನಿಸಿ)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # ಸೃಷ್ಟಿಸಲಾದ ಚಿತ್ರವನ್ನು ಪಡೆಯಿರಿ
        image_url = generation_response.data[0].url  # ಪ್ರತಿಕ್ರಿಯೆಯಿಂದ ಚಿತ್ರ URL ಅನ್ನು ತೆಗೆದುಕೊಂಡು ಬಾ
        generated_image = requests.get(image_url).content  # ಚಿತ್ರವನ್ನು ಡೌನ್‌ಲೋಡ್ ಮಾಡಿ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # ಚಿತ್ರವನ್ನು ಡಿಫಾಲ್ಟ್ ಚಿತ್ರ ವೀಕ್ಷಕದಲ್ಲಿ ಪ್ರದರ್ಶಿಸಿ
        image = Image.open(image_path)
        image.show()
    
    # ಅಪವಾದಗಳನ್ನು ಹಿಡಿಯಿರಿ
    except openai.BadRequestError as err:
        print(err)

    ```

    ```python
    # ಅಜೂರ್ ಓಪನ್‌ಎಐ ಸೇವಾ ಕ್ಲೈಯೆಂಟ್ ಅನ್ನು ಸಂರಚಿತ ಮಾಡಿ
    ```

    ```python
    # ಚಿತ್ರ ಉತ್ಪಾದನೆ ಎಪಿಐ ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸಿ
    ```

### ಚಿತ್ರ ತಯಾರಿಕಾ ಬಗ್ಗೆ ಹೆಚ್ಚಿನ ವಿವರಗಳು

ಚಿತ್ರ ತಯಾರಿಸುವ ಕೋಡ್ ಅನ್ನು ಹೆಚ್ಚಿನ ವಿವರದಲ್ಲಿ ನೋಡೋಣ:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, ಚಿತ್ರ ತಯಾರಿಸಲು ಬಳಸುವ ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಆಗಿದೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, ನಾವು "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ಎಂಬ ಪ್ರಾಂಪ್ಟ್ ಬಳಸಿ ಇದ್ದೇವೆ.
- **size**, ತಯಾರಾಗುವ ಚಿತ್ರದ ಗಾತ್ರ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, 1024x1024 ಪಿಕ್ಸೆಲ್ಸ್ ಗಾತ್ರದ ಚಿತ್ರ ತಯಾರಿಸಲಾಗಿದೆ.
- **n**, ತಯಾರಾಗುವ ಚಿತ್ರಗಳ ಸಂಖ್ಯೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, ಒಂದು ಚಿತ್ರ ತಯಾರಿಸಲಾಗಿದೆ.

ಚಿತ್ರಗಳೊಂದಿಗೆ ನೀವು ಇನ್ನಷ್ಟು ಏನು ಮಾಡಬಹುದು ಎಂಬುದನ್ನು ಮುಂದಿನ ವಿಭಾಗದಲ್ಲಿ ನೋಡೋಣ.

## ಚಿತ್ರ ತಯಾರಿಕೆಯ ಹೆಚ್ಚುವರಿ ಶಕ್ತಿಗಳು

Python ನಲ್ಲಿ ಕೆಲವು ಸಾಲುಗಳೊಂದಿಗೆ ಚಿತ್ರವನ್ನು ತಯಾರಿಸಲು ನಮಗೆ ಯೋಗ್ಯತೆ ಎಷ್ಟರಮಟ್ಟಿಗೆ ದೊರಕಿತುಂದು ಈಗವರೆಗೂ ನೋಡಿದ್ದೀರಿ. ಆದರೆ ಚಿತ್ರಗಳೊಂದಿಗೆ ಇನ್ನಷ್ಟು ಕೆಲಸ ಮಾಡಬಹುದು.

ನೀವು ಕೆಳಗಿನುವನ್ನು ಕೂಡ ಮಾಡಬಹುದು:

- **ತಿದ್ದುಪಡಿ ಮಾಡುವುದು**. ಈಗಿರುವ ಚಿತ್ರಕ್ಕೆ ಮಾಸ್ಕ್ ಮತ್ತು ಪ್ರಾಂಪ್ಟ್ ಒದಗಿಸುವ ಮೂಲಕ ನೀವು ಚಿತ್ರವನ್ನು ಬದಲಾಯಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ಕೆಲವು ಭಾಗಕ್ಕೆ ಏನಾದರು ಸೇರ್ಪಡೆ ಮಾಡಬಹುದು. ನಮ್ಮ ಬಿಡಿಯಾ ಚಿತ್ರವನ್ನು ಕಲ್ಪಿಸಿ, ನೀವು ಬನ್ನಿಗೆ ಟೋಪಿ ಸೇರ್ಪಡೆ ಮಾಡಬಹುದು. ನೀವು ಅದನ್ನು ಹೇಗೆ ಮಾಡುವುದು ಎಂದರೆ, ಚಿತ್ರ, ಮಾಸ್ಕ್ (ಬದಲಾವಣೆಯ ಭಾಗ ಗುರುತಿಸುವುದು) ಮತ್ತು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಒದಗಿಸುವ ಮೂಲಕ.

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
