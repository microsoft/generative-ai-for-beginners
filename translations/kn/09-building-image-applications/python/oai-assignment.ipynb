# ಇಮೇಜ್ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ನಿರ್ಮಿಸುವುದು (OpenAI)

LLM ಗಳಿಗೆ ಪಠ್ಯ ತಯಾರಿಕೆಗೆ ಹೆಚ್ಚುವರಿ ಮತ್ತಷ್ಟು ಅವಕಾಶಗಳಿವೆ. ನೀವು ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಕೂಡ ನಿರ್ಮಿಸಬಹುದು. ಚಿತ್ರಗಳು ಒಂದು ಮಾಧ್ಯಮವಾಗಿ ಮೆಡ್‌ಟೆಕ್, معماري, ಪ್ರವಾಸೋದ್ಯಮ, ಗೇಮ್ ಡೆವಲಪ್‌ಮೆಂಟ್, ಮಾರ್ಕೆಟಿಂಗ್ ಮತ್ತು ಇನ್ನಷ್ಟು ಕ್ಷೇತ್ರಗಳಲ್ಲಿ ಉಪಯುಕ್ತವಾಗುತ್ತವೆ. ಈ ಪಾಠದಲ್ಲಿ ನಾವು OpenAI ವೇದಿಕೆಯಲ್ಲಿ ಇಂದಿನ **GPT ಇಮೇಜ್** ಮಾದರಿಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡುತ್ತೇವೆ.

## ಕಲಿಕೆ ಗುರಿಗಳು

ಈ ಪಾಠದ ಅಂತ್ಯಕ್ಕೆ ನೀವು ಸಾಧ್ಯವಾಗಿರುತ್ತದೆ:

- ಚಿತ್ರ ತಯಾರಿಕೆ ಎಂದರೇನು ಮತ್ತು ಅದು ಎಲ್ಲಿಗೆ ಉಪಯುಕ್ತ ಎಂದು ವಿವರಿಸುವುದು.
- `gpt-image` ಮಾದರಿ ಕುಟುಂಬವನ್ನು ಹಾಗೂ ಅದು ಹಳೆಯ DALL·E ಮಾದರಿಗಳಿಂದ ಹೇಗೆ ವಿಭಿನ್ನವಾಗಿದ್ದುಕೊಳ್ಳುತ್ತದೆ ಎಂಬುದನ್ನು ತಿಳಿದುಕೊಳ್ಳುವುದು.
- ಚಿತ್ರ ತಯಾರಿಕೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಿ ಚಿತ್ರಗಳನ್ನು ಸಂಪಾದಿಸುವುದು.

## ಚಿತ್ರ ತಯಾರಿಕೆ ಎಂದರೇನು?

ಚಿತ್ರ ತಯಾರಿಕೆ ಮಾದರಿಗಳು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್‌ನಿಂದ ಚಿತ್ರಗಳನ್ನು ನಿರ್ಮಿಸುತ್ತವೆ. `gpt-image` ಮುಂತಾದ ಆಧುನಿಕ ಮಾದರಿಗಳು ತರಬೇತಿಯ ಸಮಯದಲ್ಲಿ ಪಠ್ಯ ಮತ್ತು ಚಿತ್ರಗಳ ನಡುವೆ ಸಂಬಂಧ ಕಲಿತ ತಮಗೆ, ನಂತರ ಕ್ರಮಕ್ರಮವಾಗಿ ರ್ಯಾಂಡಮ್ ಶಬ್ದವನ್ನು ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್‌ಗೆ ಹೊಂದಿಕೊಂಡಿರುವ ಚಿತ್ರವಾಗಿ ಪರಿವರ್ತಿಸುತ್ತವೆ.

ಎರಡು ಪ್ರಸಿದ್ಧ ಚಿತ್ರ ಮಾದರಿ ಕುಟುಂಬಗಳು:

- **`gpt-image` (OpenAI)** — ಈ ಪಾಠದಲ್ಲಿ ಬಳಸಲಾಗುವ ಪ್ರಸ್ತುತ ತಲೆಮಾರಿಗೆ ಸೇರಿದ ಮಾದರಿ. ಇದಕ್ಕೆ ಪಠ್ಯ-ಇಮೇಜ್ ತಯಾರಿಕೆ ಮತ್ತು ಚಿತ್ರ ಸಂಪಾದನೆ (ಮಾಸ್ಕ್‌ನೊಂದಿಗೆ ಇನ್‌ಪೇಂಟಿಂಗ್) ಸೌಲಭ್ಯವಿದೆ.
- **Midjourney** — ತನ್ನ ಸ್ವಂತ ಸೇವೆ ಮತ್ತು Discord ಆಧಾರಿತ ಕಾರ್ಯವಿಧಾನವನ್ನು ಹೊಂದಿರುವ ಜನಪ್ರಿಯ ಮೂರನೇ ಪಕ್ಷದ ಮಾದರಿ.

> ಹಳೆಯ OpenAI ಚಿತ್ರ ಮಾದರಿಗಳು — **DALL·E 2** ಮತ್ತು **DALL·E 3** — ಪರಂಪರೆ ಮಾದರಿಗಳು. DALL·E 3 ಹೊಸ ನಿಯೋಜನೆಗಳಿಗೆ ಲಭ್ಯವಿಲ್ಲ, ಮತ್ತು `create_variation` ಮುಂತಾದ ವೈಶಿಷ್ಟ್ಯಗಳು ಕೇವಲ DALL·E 2 ಯಲ್ಲಿ ಇದ್ದವು. ಹೊಸ ಅಪ್ಲಿಕೇಶನ್‌ಗಳಿಗೆ `gpt-image` ಮಾದರಿಗಳನ್ನು ಉಪಯೋಗಿಸಿ.

> ** ಮಹತ್ವಪೂರ್ಣ:** `gpt-image` ಮಾದರಿಗಳು ತಯಾರಿಸಿದ ಚಿತ್ರವನ್ನು URL ರೂಪದಲ್ಲಿ ಅಲ್ಲ, ಬದಲಿಗೆ **ಬೇಸ್64** (`b64_json`) ರೂಪದಲ್ಲಿ ರಿಟರ್ನ್ ಮಾಡುತ್ತವೆ. ನಿಮ್ಮ ಕೋಡ್ ಬೇಸ್64 ಸ್ಟ್ರಿಂಗ್ ಅನ್ನು ಬೈಟ್ಸ್‌ ಆಗಿ ಡೀಕೋಡ್ ಮಾಡಿ ಅದನ್ನು ಉಳಿಸುತ್ತದೆ — ಡೌನ್‌ಲೋಡ್ ಮಾಡಲು ಚಿತ್ರ URL ಇರುತ್ತದೆಂದು ಇಲ್ಲ.


## ನಿಮ್ಮ ಮೊದಲ ಚಿತ್ರ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು

ಹೀಗಾಗಿ ಚಿತ್ರ ಉತ್ಪಾದನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲು ಯಾವ್ದೇನು ಬೇಕಾಗುತ್ತದೆ? ನಿಮಗೆ ಕೆಳಗಿನ ಗ್ರಂಥಾಲಯಗಳು ಬೇಕಾಗಿವೆ:

- **python-dotenv**, ನಿಮ್ಮ ರಹಸ್ಯಗಳನ್ನು ಕೋಡ್‌ ರಿಂದ ದೂರವಾಗಿರುವ *.env* ಫೈಲ್‌ನಲ್ಲಿ ಇಡಿಯಲು ಈ ಗ್ರಂಥಾಲಯವನ್ನು ಶಿಫಾರಸು ಮಾಡಲಾಗುತ್ತದೆ.
- **openai**, ನೀವು OpenAI API ಜೊತೆಗೆ ಸಂವಹನಿಸಲು ಈ ಗ್ರಂಥಾಲಯವನ್ನು ಬಳಸುವಿರಿ.
- **pillow**, Python ನಲ್ಲಿ ಚಿತ್ರಗಳ ಜೊತೆ ಕೆಲಸ ಮಾಡಲು.


1. ಕೆಳಗಿನ ವಿಷಯವನ್ನು ಒಳಗೊಂಡ *.env* ಎಂಬ ಫೈಲ್ ಸೃಷ್ಟಿಸಿ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. ಮೇಲಿನ ಲೈಬ್ರರಿಗಳನ್ನು *requirements.txt* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಈ ರೀತಿ ಸಂಗ್ರಹಿಸಿ:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. ನಂತರ, ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ಸೃಷ್ಟಿಸಿ ಮತ್ತು ಲೈಬ್ರರಿಗಳನ್ನು ಸ್ಥಾಪಿಸಿ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> ವಿಂಡೋಸ್‌ಗೆ, ನಿಮ್ಮ ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಲು ಮತ್ತು ಸಕ್ರಿಯಗೊಳಿಸಲು ಕೆಳಗಿನ ಕಮಾಂಡ್ಗಳನ್ನು ಬಳಸಿ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಕೆಳಗಿನ ಕೋಡನ್ನು ಸೇರಿಸಿ:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv ಅನ್ನು ಆಮದುಮಾಡಿ
    dotenv.load_dotenv()

    # OpenAI ವಸ್ತುವನ್ನು ರಚಿಸಿ (.env ನಿಂದ OPENAI_API_KEY ಅನ್ನು ಓದುತ್ತದೆ)
    client = openai.OpenAI()


    try:
        # ಚಿತ್ರ ಉತ್ಪಾದನೆ API ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸಿ
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಾಠವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
            size='1024x1024',
            n=1
        )
        # ಸಂಗ್ರಹಿಸಿದ ಚಿತ್ರಕ್ಕಾಗಿ ಡೈರೆಕ್ಟರಿಯನ್ನು ಸೆಟ್ ಮಾಡಿ
        image_dir = os.path.join(os.curdir, 'images')

        # ಡೈರೆಕ್ಟರಿ ಇಲ್ಲದಿದ್ದಲ್ಲಿ, ಅದನ್ನು ರಚಿಸಿ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ಚಿತ್ರದ ಮಾರ್ಗವನ್ನು ಪ್ರಾರಂಭಿಸಿ (ಫೈಲ್ ಪ್ರಕಾರ png ಆಗಿರಬೇಕು ಎಂದು ಗಮನಿಸಿ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image ಮಾದರಿಗಳು ಚಿತ್ರವನ್ನು URL ಅಲ್ಲದೆ;base64 (b64_json) ರೂಪದಲ್ಲಿ ನೀಡುತ್ತವೆ
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ಡೀಫಾಲ್ಟ್ ಚಿತ್ರ ವೀಕ್ಷಕದಲ್ಲಿ ಚಿತ್ರವನ್ನು ತೋರಿಸಿ
        image = Image.open(image_path)
        image.show()

    # ಅಪವಾದಗಳನ್ನು ಹಿಡಿಯಿರಿ
    except openai.BadRequestError as err:
        print(err)

    ```

ಈ ಕೋಡ್ ಅನ್ನು ವಿವರಿಸೋಣ:

- ಮೊದಲು, ನಾವು ಬೇಕಾಗುವ ಗ್ರಂಥಾಲಯಗಳನ್ನು ಎ.importಮೌಮ್ ಮಾಡುತ್ತೇವೆ, ಇದರಲ್ಲಿ OpenAI ಗ್ರಂಥಾಲಯ, dotenv ಗ್ರಂಥಾಲಯ, base64 ಮೋಡುಲ್, ಮತ್ತು Pillow ಗ್ರಂಥಾಲಯ ಸೇರಿವೆ.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- ಅದಾದ ಮೇಲೆ, ನಾವು ಕ್ಲೈಂಟ್ ಅನ್ನು ರಚಿಸುತ್ತೇವೆ, ಇದು API ಕೀ ಅನ್ನು ನಿಮ್ಮ ``.env`` ಫೈಲ್‌ನಿಂದ ಓದುವದು.

    ```python
    # OpenAI ವಸ್ತುವನ್ನು ರಚಿಸಿ
    client = openai.OpenAI()
    ```

- ನಂತರ, ನಾವು ಚಿತ್ರವನ್ನು ರಚಿಸುತ್ತೇವೆ:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ಇಲ್ಲಿ ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯದ ಸೇರ್ಪಡೆ ಮಾಡಿ
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` ಮಾದರಿಗಳು ಚಿತ್ರವನ್ನು **base64** ಸ್ಟ್ರಿಂಗ್ ಆಗಿ `data[0].b64_json` ನಲ್ಲಿ ನೀಡುತ್ತವೆ. ನಾವು ಅದನ್ನು ಬೈಟ್ಸ್ ಗೆ ಡಿಕೋಡ್ ಮಾಡಿ ಫೈಲ್ ಗೆ ಬರೆಯುತ್ತೇವೆ — ಡೌನ್‌ಲೋಡ್ ಮಾಡಲು ಯಾವುದೇ URL ಇರುವುದಿಲ್ಲ.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- ಕೊನೆಯದಾಗಿ, ನಾವು ಚಿತ್ರವನ್ನು ತೆರೆದು ಸಾಮಾನ್ಯ ಚಿತ್ರ ವೀಕ್ಷಕ ಬಳಸಿ ಅದನ್ನು ಪ್ರದರ್ಶಿಸುತ್ತೇವೆ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ಚಿತ್ರ ರಚಿಸುವ ಕುರಿತು ಹೆಚ್ಚಿನ ವಿವರಗಳು

`images.generate` ನ ಪ್ಯಾರಾಮೀಟರ್ಗಳಿಗೆ ಒಂದು ನೋಟ ಹಾಕೋಣ:

- **model**, ಬಳಸುವ ಚಿತ್ರ ಮಾದರಿ (ಉದಾಹರಣೆಗೆ `gpt-image-1`).
- **prompt**, ಚಿತ್ರ ರಚಿಸಲು ಬಳಸುವ ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್. ಇಲ್ಲಿ ಅದು "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, ರಚಿಸಲಾದ ಚಿತ್ರದ ಗಾತ್ರ (ಉದಾಹರಣೆಗೆ `1024x1024`, `1536x1024`, `1024x1536`, ಅಥವಾ `"auto"`).
- **n**, ರಚಿಸಲಾದ ಚಿತ್ರಗಳ ಸಂಖ್ಯೆ. ಇಲ್ಲಿ ನಾವು ಒಂದು ಚಿತ್ರವನ್ನು ರಚಿಸುತ್ತೇವೆ.

> ಚಿತ್ರ ಮಾದರಿಗಳು `temperature` ಎಂಬ ಪ್ಯಾರಾಮೀಟರ್ ತೆಗೆದುಕೊಳ್ಳುವುದಿಲ್ಲ — ಅದು ಪಠ್ಯ ರಚನೆ ನಿಯಂತ್ರಣ. ವೈವಿಧ್ಯತೆಗೆ API ಅನ್ನು ಮತ್ತೆ ಕರೆಮಾಡಿ; ವೈವಿಧ್ಯತೆ ಕಡಿಮೆ ಮಾಡಲು, ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ವಿಶೇಷಗೊಳಿಸಿ.

## ಚಿತ್ರ ರಚನೆಯ ಹೆಚ್ಚುವರಿ ಸಾಮರ್ಥ್ಯಗಳು

ನೀವು ಕೆಲವು ಪೈಥಾನ್ ಸಾಲುಗಳೊಂದಿಗೆ ಹೇಗೆ ಚಿತ್ರವನ್ನು ರಚಿಸಬಹುದು ಎಂದು ನೋಡಿದ್ದೀರಿ. `gpt-image` ಮಾದರಿಗಳು ಸಲ್ಲಿರುವ ಚಿತ್ರವನ್ನು **ಸಂಪಾದಿಸಬಹುದಾಗಿದೆ**. ಚಿತ್ರವನ್ನು ಒದಗಿಸಿ, ಐಚ್ಛಿಕ **ಮಾಸ್ಕ್** (ಬದಲಾಯಿಸಲು ಇರುವ ಭಾಗವನ್ನು ಸೂಚಿಸುತ್ತದೆ) ಮತ್ತು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ನೀಡುವುದರಿಂದ ಚಿತ್ರದಲ್ಲಿ ಭಾಗವನ್ನು ಬದಲಾಯಿಸಬಹುದು — ಉದಾಹರಣೆಗೆ, ನಮ್ಮ ಹುಳುಬಿದ್ದ ಬನಿಗೆ ಟೊಪಿ ಸೇರಿಸಬಹುದು.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ತಿದ್ದುಪಾಡುಗಳು ಕೂಡ base64 ಆಗಿ ಹಿಂತಿರುಗಿಸಲಾಗುತ್ತದೆ
edited_image = base64.b64decode(response.data[0].b64_json)
```

ಮೂಲ ಚಿತ್ರದಲ್ಲಿ ಕೇವಲ ಖರಗಿ ಇದೆ; ಅಂತಿಮ ಚಿತ್ರದಲ್ಲಿ ಖರಗಿಗೆ ಟೊಪಿ ಇದೆ.
