# ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಾಣ

LLM ಗಳು ಪಠ್ಯ ರಚನೆಗೆ ಮಾತ್ರವಲ್ಲ. ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸುವುದೂ ಸಾಧ್ಯ. ಚಿತ್ರಗಳನ್ನು ಒಂದು ಮಾದರಿಯಾಗಿ ಹೊಂದಿರುವುದು ಮೆಡ್‌ಟೆಕ್, ವಾಸ್ತುಶಿಲ್ಪ, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಮತ್ತು ಇನ್ನಷ್ಟು ಕ್ಷೇತ್ರಗಳಲ್ಲಿ ಬಹಳ ಉಪಯುಕ್ತವಾಗಬಹುದು. ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನಾವು ಎರಡು ಅತ್ಯಂತ ಜನಪ್ರಿಯ ಚಿತ್ರ ರಚನೆ ಮಾದರಿಗಳನ್ನು, DALL-E ಮತ್ತು Midjourney ಅನ್ನು ನೋಡೋಣ.

## ಪರಿಚಯ

ಈ ಪಾಠದಲ್ಲಿ, ನಾವು ಒಳಗೊಂಡಿರುವುದು:

- ಚಿತ್ರ ರಚನೆ ಮತ್ತು ಅದು ಯಾಕೆ ಉಪಯುಕ್ತವಾಗಿದೆ.
- DALL-E ಮತ್ತು Midjourney, ಅವು ಏನು ಮತ್ತು ಅವು ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತವೆ.
- ನೀವು ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ಅನ್ನು ಹೇಗೆ ನಿರ್ಮಿಸುವಿರಿ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ನಂತರ, ನೀವು ಸಾಧ್ಯವಾಗುವುದು:

- ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು.
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಗಾಗಿ ಮೆಟಾ ಪ್ರಾಂಪ್ಟ್‌ಗಳೊಂದಿಗೆ ಗಡಿಗಳನ್ನು ನಿರ್ಧರಿಸುವುದು.
- DALL-E ಮತ್ತು Midjourney ಜೊತೆಗೆ ಕೆಲಸ ಮಾಡುವುದು.

## ಯಾಕೆ ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಬೇಕು?

ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್‌ಗಳು ಜನರೇಟಿವ್ AI ಸಾಮರ್ಥ್ಯಗಳನ್ನು ಅನ್ವೇಷಿಸಲು ಉತ್ತಮ ಮಾರ್ಗವಾಗಿದೆ. ಉದಾಹರಣೆಗೆ, ಅವುಗಳನ್ನು ಬಳಸಬಹುದು:

- **ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಸಂಶ್ಲೇಷಣೆ**. ನೀವು ವಿವಿಧ ಬಳಕೆ ಪ್ರಕರಣಗಳಿಗೆ ಚಿತ್ರಗಳನ್ನು ರಚಿಸಬಹುದು, ಉದಾಹರಣೆಗೆ ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಚಿತ್ರ ಸಂಶ್ಲೇಷಣೆ.

- **ವಿವಿಧ ಉದ್ಯಮಗಳಿಗೆ ಅನ್ವಯಿಸುವುದು**. ಅವು ಮೆಡ್‌ಟೆಕ್, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಮತ್ತು ಇನ್ನಷ್ಟು ಉದ್ಯಮಗಳಿಗೆ ಚಿತ್ರಗಳನ್ನು ರಚಿಸಲು ಬಳಸಬಹುದು.

## DALL-E ಮತ್ತು Midjourney ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತವೆ

ಮೊದಲು, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ಒಂದು ಜನರೇಟಿವ್ AI ಮಾದರಿ, ಇದು ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್ ವಾಸ್ತುಶಿಲ್ಪದ ಮೇಲೆ ಆಧಾರಿತವಾಗಿದ್ದು, *ಆಟೋರೆಗ್ರೆಸಿವ್ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್* ಅನ್ನು ಹೊಂದಿದೆ.

*ಆಟೋರೆಗ್ರೆಸಿವ್ ಟ್ರಾನ್ಸ್‌ಫಾರ್ಮರ್* ಒಂದು ಮಾದರಿ ಪಠ್ಯ ವಿವರಣೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಹೇಗೆ ರಚಿಸುತ್ತದೆ ಎಂದು ನಿರ್ಧರಿಸುತ್ತದೆ, ಇದು ಒಂದೊಂದು ಪಿಕ್ಸೆಲ್ ಅನ್ನು ಕ್ರಮವಾಗಿ ರಚಿಸಿ, ನಂತರ ರಚಿಸಿದ ಪಿಕ್ಸೆಲ್‌ಗಳನ್ನು ಬಳಸಿ ಮುಂದಿನ ಪಿಕ್ಸೆಲ್ ಅನ್ನು ರಚಿಸುತ್ತದೆ. ನ್ಯೂರಲ್ ನೆಟ್‌ವರ್ಕ್‌ನ ಅನೇಕ ಪದರಗಳ ಮೂಲಕ ಪಾಸ್ ಆಗಿ, ಚಿತ್ರ ಪೂರ್ಣಗೊಳ್ಳುವವರೆಗೆ.

ಈ ಪ್ರಕ್ರಿಯೆಯಿಂದ, DALL-E ರಚಿಸುವ ಚಿತ್ರದಲ್ಲಿ ಗುಣಲಕ್ಷಣಗಳು, ವಸ್ತುಗಳು, ಲಕ್ಷಣಗಳು ಮತ್ತು ಇನ್ನಷ್ಟು ನಿಯಂತ್ರಣ ಮಾಡುತ್ತದೆ. ಆದಾಗ್ಯೂ, DALL-E 2 ಮತ್ತು 3 ರಲ್ಲಿ ರಚಿಸಿದ ಚಿತ್ರಗಳ ಮೇಲೆ ಹೆಚ್ಚು ನಿಯಂತ್ರಣವಿದೆ.


## ನಿಮ್ಮ ಮೊದಲ ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು

ಆಗ ಚಿತ್ರ ರಚನೆ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲು ಏನು ಬೇಕಾಗುತ್ತದೆ? ನಿಮಗೆ ಕೆಳಗಿನ ಗ್ರಂಥಾಲಯಗಳು ಬೇಕಾಗಿವೆ:

- **python-dotenv**, ನಿಮ್ಮ ರಹಸ್ಯಗಳನ್ನು ಕೋಡ್‌ನಿಂದ ದೂರವಿರುವ *.env* ಫೈಲ್‌ನಲ್ಲಿ ಇಡಲು ಈ ಗ್ರಂಥಾಲಯವನ್ನು ಬಳಸಲು ನಿಮಗೆ ಶಿಫಾರಸು ಮಾಡಲಾಗಿದೆ.
- **openai**, ಈ ಗ್ರಂಥಾಲಯವನ್ನು ನೀವು OpenAI API ಜೊತೆಗೆ ಸಂವಹನ ಮಾಡಲು ಬಳಸುತ್ತೀರಿ.
- **pillow**, Python ನಲ್ಲಿ ಚಿತ್ರಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು.
- **requests**, HTTP ವಿನಂತಿಗಳನ್ನು ಮಾಡಲು ಸಹಾಯ ಮಾಡಲು.

1. ಕೆಳಗಿನ ವಿಷಯವನ್ನು ಹೊಂದಿರುವ *.env* ಫೈಲ್ ಅನ್ನು ರಚಿಸಿ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```


1. ಮೇಲಿನ ಗ್ರಂಥಾಲಯಗಳನ್ನು *requirements.txt* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಈ ರೀತಿಯಾಗಿ ಸಂಗ್ರಹಿಸಿ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. ನಂತರ, ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಿ ಮತ್ತು ಗ್ರಂಥಾಲಯಗಳನ್ನು ಸ್ಥಾಪಿಸಿ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> ವಿಂಡೋಸ್‌ಗಾಗಿ, ನಿಮ್ಮ ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಲು ಮತ್ತು ಸಕ್ರಿಯಗೊಳಿಸಲು ಕೆಳಗಿನ ಆಜ್ಞೆಗಳನ್ನು ಬಳಸಿ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ಎಂಬ ಫೈಲ್‌ನಲ್ಲಿ ಕೆಳಗಿನ ಕೋಡ್ ಅನ್ನು ಸೇರಿಸಿ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ಅನ್ನು ಆಮದುಮಾಡಿ
    dotenv.load_dotenv()

    # OpenAI ವಸ್ತುವನ್ನು ರಚಿಸಿ
    client = OpenAI()


    try:
        # ಚಿತ್ರ ರಚನೆ API ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸಿ
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
            size='1024x1024',
            n=1
        )
        # ಸಂಗ್ರಹಿಸಲಾದ ಚಿತ್ರಕ್ಕಾಗಿ ಡೈರೆಕ್ಟರಿಯನ್ನು ಸೆಟ್ ಮಾಡಿ
        image_dir = os.path.join(os.curdir, 'images')

        # ಡೈರೆಕ್ಟರಿ ಇಲ್ಲದಿದ್ದರೆ, ಅದನ್ನು ರಚಿಸಿ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ಚಿತ್ರ ಮಾರ್ಗವನ್ನು ಪ್ರಾರಂಭಿಸಿ (ಫೈಲ್ ಪ್ರಕಾರ png ಆಗಿರಬೇಕು ಎಂದು ಗಮನಿಸಿ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # ರಚಿಸಲಾದ ಚಿತ್ರವನ್ನು ಪಡೆಯಿರಿ
        print(generation_response)

        image_url = generation_response.data[0].url # ಪ್ರತಿಕ್ರಿಯೆಯಿಂದ ಚಿತ್ರ URL ಅನ್ನು ಹೊರತೆಗೆಯಿರಿ
        generated_image = requests.get(image_url).content  # ಚಿತ್ರವನ್ನು ಡೌನ್‌ಲೋಡ್ ಮಾಡಿ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ಡೀಫಾಲ್ಟ್ ಚಿತ್ರ ವೀಕ್ಷಕದಲ್ಲಿ ಚಿತ್ರವನ್ನು ಪ್ರದರ್ಶಿಸಿ
        image = Image.open(image_path)
        image.show()

    # исключенияಗಳನ್ನು ಹಿಡಿಯಿರಿ
    except client.error.InvalidRequestError as err:
        print(err)

    ```

### ಚಿತ್ರ ರಚನೆ ಕುರಿತು ಹೆಚ್ಚಿನ ವಿವರಗಳು

ಚಿತ್ರವನ್ನು ರಚಿಸುವ ಕೋಡ್ ಅನ್ನು ಹೆಚ್ಚಿನ ವಿವರದಲ್ಲಿ ನೋಡೋಣ:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
        size='1024x1024',
        n=1
    )
```

- **prompt**, ಚಿತ್ರವನ್ನು ರಚಿಸಲು ಬಳಸುವ ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್. ಈ ಪ್ರಕರಣದಲ್ಲಿ, ನಾವು "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ಎಂಬ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಬಳಸುತ್ತಿದ್ದೇವೆ.
- **size**, ರಚಿಸಲಾದ ಚಿತ್ರದ ಗಾತ್ರ. ಈ ಪ್ರಕರಣದಲ್ಲಿ, ನಾವು 1024x1024 ಪಿಕ್ಸೆಲ್ ಗಾತ್ರದ ಚಿತ್ರವನ್ನು ರಚಿಸುತ್ತಿದ್ದೇವೆ.
- **n**, ರಚಿಸಲಾದ ಚಿತ್ರಗಳ ಸಂಖ್ಯೆ. ಈ ಪ್ರಕರಣದಲ್ಲಿ, ನಾವು ಎರಡು ಚಿತ್ರಗಳನ್ನು ರಚಿಸುತ್ತಿದ್ದೇವೆ.

ಚಿತ್ರಗಳೊಂದಿಗೆ ನೀವು ಇನ್ನೂ ಹೆಚ್ಚಿನ ಕಾರ್ಯಗಳನ್ನು ಮಾಡಬಹುದು, ಅವುಗಳನ್ನು ಮುಂದಿನ ವಿಭಾಗದಲ್ಲಿ ನಾವು ಚರ್ಚಿಸುವೆವು.

## ಚಿತ್ರ ರಚನೆಯ ಹೆಚ್ಚುವರಿ ಸಾಮರ್ಥ್ಯಗಳು

ನೀವು ಈಗಾಗಲೇ ಪೈಥಾನ್‌ನಲ್ಲಿ ಕೆಲವು ಸಾಲುಗಳನ್ನು ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸುವುದನ್ನು ನೋಡಿದ್ದೀರಿ. ಆದರೆ, ಚಿತ್ರಗಳೊಂದಿಗೆ ನೀವು ಇನ್ನೂ ಹೆಚ್ಚಿನ ಕಾರ್ಯಗಳನ್ನು ಮಾಡಬಹುದು.

ನೀವು ಕೆಳಗಿನ ಕಾರ್ಯಗಳನ್ನು ಸಹ ಮಾಡಬಹುದು:

- **ತಿದ್ದುಪಡಿ ಮಾಡುವುದು**. ಈಗಾಗಲೇ ಇರುವ ಚಿತ್ರಕ್ಕೆ ಮಾಸ್ಕ್ ಮತ್ತು ಪ್ರಾಂಪ್ಟ್ ಒದಗಿಸುವ ಮೂಲಕ, ನೀವು ಚಿತ್ರವನ್ನು ಬದಲಾಯಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ಚಿತ್ರದ ಒಂದು ಭಾಗಕ್ಕೆ ಏನಾದರೂ ಸೇರಿಸಬಹುದು. ನಮ್ಮ ಬನ್ನಿ ಚಿತ್ರವನ್ನು ಕಲ್ಪಿಸಿ, ನೀವು ಬನ್ನಿಗೆ ಟೋಪಿ ಸೇರಿಸಬಹುದು. ನೀವು ಅದನ್ನು ಹೇಗೆ ಮಾಡುತ್ತೀರಿ ಎಂದರೆ, ಚಿತ್ರ, ಮಾಸ್ಕ್ (ಬದಲಾವಣೆಗಾಗಿ ಭಾಗವನ್ನು ಗುರುತಿಸುವುದು) ಮತ್ತು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು ಒದಗಿಸುವ ಮೂಲಕ.

    ```python
    response = openai.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ಮೂಲ ಚಿತ್ರದಲ್ಲಿ ಕೇವಲ ಮೊಲ ಮಾತ್ರ ಇರುತ್ತದೆ ಆದರೆ ಅಂತಿಮ ಚಿತ್ರದಲ್ಲಿ ಮೊಲದ ಮೇಲೆ ಟೋಪಿ ಇರುತ್ತದೆ.
    
- **ವೈವಿಧ್ಯಗಳನ್ನು ರಚಿಸುವುದು**. ಆಲೋಚನೆ ಎಂದರೆ ನೀವು ಈಗಾಗಲೇ ಇರುವ ಚಿತ್ರವನ್ನು ತೆಗೆದುಕೊಂಡು ಅದರ ವೈವಿಧ್ಯಗಳನ್ನು ರಚಿಸುವಂತೆ ಕೇಳುವುದು. ವೈವಿಧ್ಯವನ್ನು ರಚಿಸಲು, ನೀವು ಚಿತ್ರ ಮತ್ತು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಒದಗಿಸಿ ಕೆಳಗಿನಂತೆಯೇ ಕೋಡ್ ಬರೆಯುತ್ತೀರಿ:

    ```python
    response = openai.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
