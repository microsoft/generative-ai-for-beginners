# ಚಿತ್ರ ನಿರ್ಮಾಣ ಅನ್ವಯಿಕೆಯನ್ನು ನಿರ್ಮಿಸುವುದು 

LLMಗಳು ಪಠ್ಯ ನಿರ್ಮಾಣಕ್ಕಿಂತ ಹೆಚ್ಚಿದೆ. ಪಠ್ಯ ವರ್ಣನೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸುವುದು ಸಹ ಸಾಧ್ಯವಾಗಿದೆ. ಚಿತ್ರಗಳು ಒಂದು ಮಾದರಿಯಾಗಿರುವುದು ಮೆಡ್‌ಟೆಕ್, ವಾಸ್ತುಶಿಲ್ಪ, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಮತ್ತು ಇನ್ನಷ್ಟು ಬೇರೆ ಕ್ಷೇತ್ರಗಳಲ್ಲಿ ಅತ್ಯಂತ ಉಪಯುಕ್ತವಾಗಬಹುದು. ಈ ಅಧ್ಯಾಯದಲ್ಲಿ, ನಾವು ಅತ್ಯಂತ ಜನಪ್ರಿಯ ಚಿತ್ರ ಉತ್ಪಾದನಾ ಮಾದರಿಗಳಾದ DALL-E ಮತ್ತು Midjourney ಗಳನ್ನು ನೋಡುತ್ತೇವೆ.

## ಪರಿಚಯ 

ಈ ಪಾಠದಲ್ಲಿ, ನಾವು ಚರ್ಚಿಸುವುದು: 

- ಚಿತ್ರ ನಿರ್ಮಾಣ ಮತ್ತು ಅದಾ ಬಳಸುವ ಉಪಯುಕ್ತತೆ.
- DALL-E ಮತ್ತು Midjourney, ಅವುಗಳೇನು ಮತ್ತು ಅವು ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತವೆ.
- ಚಿತ್ರ ಉತ್ಪಾದನಾ ಆಪ್ ಅನ್ನು ನೀವು ಹೇಗೆ ನಿರ್ಮಿಸುವಿರಿ.

## ಕಲಿಕೆಯ ಗುರಿಗಳು 

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ಮೇಲೆ, ನೀವು ಸಾಧ್ಯವಿರುವುದು: 

- ಚಿತ್ರ ನಿರ್ಮಾಣ ಅನ್ವಯಿಕೆಯನ್ನು ನಿರ್ಮಿಸುವುದು.
- ಮೆಟಾ ಪ್ರಾಂಪ್ಟ್ ಗಳೊಂದಿಗೆ ನಿಮ್ಮ ಅನ್ವಯಿಕೆಗೆ ಗಡಿಗಳನ್ನು ನಿರ್ಧರಿಸುವುದು.
- DALL-E ಮತ್ತು Midjourney ಜೊತೆಗೆ ಕೆಲಸ ಮಾಡುವುದು.

## ಚಿತ್ರ ಉತ್ಪಾದನಾ ಅನ್ವಯಿಕೆಯನ್ನು ಏಕೆ ನಿರ್ಮಿಸಬೇಕು?

ಚಿತ್ರ ಉತ್ಪಾದನಾ ಅನ್ವಯಿಕೆಗಳು ಜನರೇಟಿವ್ AI ಶಕ್ತಿಯನ್ನು ಅನ್ವೇಷಿಸಲು ಅತ್ಯುತ್ತಮ ಮಾರ್ಗಗಳಾಗಿವೆ. ಉದಾಹರಣೆಗೆ, ಅವುಗಳನ್ನು ಬಳಸಬಹುದು:  

- **ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಸಂಶ್ಲೇಷಣೆ**. ಚಿತ್ರ ಸಂಪಾದನೆ ಮತ್ತು ಸಂಶ್ಲೇಷಣೆ ಮೊದಲಾದ ಬಗೆಬಗೆಯ ಬಳಕೆಗೆ ಚಿತ್ರಗಳನ್ನು ರಚಿಸಬಹುದು.  

- **ವಿವಿಧ ಉದ್ಯಮಗಳಿಗೆ ಅನ್ವಯಿಸಬಹುದು**. ಮೆಡ್‌ಟೆಕ್, ಪ್ರವಾಸೋದ್ಯಮ, ಆಟ ಅಭಿವೃದ್ಧಿ ಮತ್ತು ಇನ್ನಷ್ಟು ಉದ್ಯಮಗಳಿಗೆ ಚಿತ್ರ ರಚಿಸಲು ಬಳಸಬಹುದು.

## DALL-E ಮತ್ತು Midjourney ಏನು? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) ಮತ್ತು [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) ಇವು ಎರಡು ಅತ್ಯಂತ ಜನಪ್ರಿಯ ಚಿತ್ರ ಉತ್ಪಾದನಾ ಮಾದರಿಗಳಾಗಿವೆ; ಇವು ಪ್ರಾಂಪ್ಟ್ ಗಳ ಮೂಲಕ ಚಿತ್ರಗಳನ್ನು ರಚಿಸಲು ಅವಕಾಶ ಕೊಡುತ್ತವೆ.

### DALL-E

DALL-E ನಿಂದ ಆರಂಭಿಸೋಣ, ಇದು ಪಠ್ಯ ವರ್ಣನೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸುವ ಜನರೇಟಿವ್ AI ಮಾದರಿ. 

> [DALL-E ಎರಡು ಮಾದರಿಗಳ ಸಂಯೋಜನೆ CLIP ಮತ್ತು ಡಿಫ್ಯೂಸ್ಡ್ ಅಟೆಂಟನ್ ಆಗಿದೆ](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **CLIP**, ಚಿತ್ರಗಳು ಮತ್ತು ಪಠ್ಯಗಳಿಂದ ಸಂಖ್ಯೆ ರೂಪ ಪ್ರತಿನಿಧಿಗಳನ್ನು ರಚಿಸುವ ನಿರ್ದಿಷ್ಟ ಮಾದರಿ.  

- **ಡಿಫ್ಯೂಸ್ಡ್ ಅಟೆಂಟನ್**, ಎಂಬ ಮಾದರಿ ಚಿತ್ರಗಳನ್ನು ಪ್ರತಿನಿಧಿಗಳಿಂದ ನಿರ್ಮಿಸುತ್ತದೆ. DALL-E ಚಿತ್ರ ಮತ್ತು ಪಠ್ಯದ ಡೇಟಾಸೆಟ್ ಮೇಲೆ ತರಬೇತಿ ಪಡೆದಿದ್ದು, ಪಠ್ಯ ವರ್ಣನೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸಲು ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ನಾಯಿ ಟೋಪಿ ಹಾಕಿದ್ದ ಚಿತ್ರ ಅಥವಾ ಮೊಹಾಕ್ ಇಡುವ ನಾಯಿ ಚಿತ್ರಗಳನ್ನು DALL-E ರಚಿಸಬಹುದು. 

### Midjourney
 
Midjourney DALL-E ಮಾದರಿಯನ್ನು ಹೋಲುವಂತೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ, ಅದು ಪಠ್ಯ ಪ್ರಾಂಪ್ಟ್ ಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ರಚಿಸುತ್ತದೆ. Midjourney “ಟೋಪಿ ಹಾಕಿದ ನಾಯಿ” ಅಥವಾ “ಮೊಹಾಕ್ ಇರುವ ನಾಯಿ” ಪ್ರಾಂಪ್ಟ್ ಗಳನ್ನು ಬಳಸಿಕೊಂಡು ಚಿತ್ರಗಳನ್ನು ರಚಿಸಬಹುದು.

## DALL-E ಮತ್ತು Midjourney ಹೇಗೆ ಕೆಲಸ ಮಾಡುತ್ತವೆ 

ಮೊದಲಾಗಿ, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ಒಂದು ಜನರೇಟಿವ್ AI ಮಾದರಿ transformer ವಾಸ್ತುಶಿಲ್ಪವನ್ನು ಆಧರಿಸಿದೆ, ಇದರಲ್ಲಿ *autoregressive transformer* ಇದೆ.  

ಒಂದು *autoregressive transformer* ಎನ್ನುವುದು ಪಠ್ಯ ವರ್ಣನೆಗಳಿಂದ ಚಿತ್ರಗಳನ್ನು ಹೇಗೆ ರಚಿಸುವುದನ್ನು ನಿರ್ಧರಿಸುತ್ತದೆ, ಅದು ಒಂದು ಬಾರಿ ಒಂದು ಪಿಕ್ಸೆಲ್ ರಚಿಸಿ ನಂತರ ಆಗಾಗ ಮುಂದಿನ ಪಿಕ್ಸೆಲ್ ಅನ್ನು ರಚಿಸುವ ವಿಧಾನವನ್ನು ಅನುಸರಿಸುತ್ತದೆ. ಇದು ಅನೆಕ ಪದರಗಳಿಂದ ರಜತಹಸ್ತ network ಮೂಲಕ ಪಾಸ್ ಆಗುತ್ತದೆ, ಚಿತ್ರ ಪೂರ್ಣಗೊಳ್ಳುವವರೆಗೆ.  

ಈ ಪ್ರಕ್ರಿಯೆಯಿಂದ DALL-E ಚಿತ್ರದಲ್ಲಿ ಗುಣಲಕ್ಷಣಗಳು, ವಸ್ತುವುಗಳು, ಲಕ್ಷಣಗಳು ಮತ್ತು ಇನ್ನಷ್ಟು ನಿಯಂತ್ರಣವನ್ನು ಹೊಂದಿರುತ್ತದೆ. ಆದರೆ, DALL-E 2 ಮತ್ತು 3 ರವರು ರಚಿಸಿದ ಚಿತ್ರವನ್ನು ಇನ್ನಷ್ಟು ನಿಯಂತ್ರಿಸುತ್ತಾರೆ, 


## ನಿಮ್ಮ ಮೊದಲ ಚಿತ್ರ ತಯಾರಿಸುವ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸುವುದು

ಹಾಗಾದರೆ ಚಿತ್ರ ತಯಾರಿಸುವ ಅಪ್ಲಿಕೇಶನ್ ನಿರ್ಮಿಸಲು ಏನು ಬೇಕಾಗುತ್ತದೆ? ನಿಮಗೆ ಕೆಳಗಿನ ಲೈಬ್ರರಿಗಳು ಬೇಕಾಗುತ್ತವೆ:

- **python-dotenv**, ನಿಮ್ಮ ರಹಸ್ಯಗಳನ್ನು ಕಾಳಜಿ ವಹಿಸಲು ಮತ್ತು ಕೋಡ್‌ನಿಂದ ದೂರವಿರುವ *.env* ಫೈಲ್‌ನಲ್ಲಿ ಇಡಲು ಈ ಲೈಬ್ರರಿಯನ್ನು ಬಳಸಲು ಶಿಫಾರಸು ಮಾಡಲಾಗಿದೆ.
- **openai**, OpenAI API ಜೊತೆ ಸಂವಹನ ಮಾಡಲು ನೀವು ಬಳಸುವ ಲೈಬ್ರರಿ.
- **pillow**, Python ನಲ್ಲಿ ಚಿತ್ರಗಳೊಂದಿಗೆ ಕೆಲಸ ಮಾಡಲು.
- **requests**, HTTP ವಿನಂತಿಗಳನ್ನು ಮಾಡಲು ಸಹಾಯ ಮಾಡಲು.


1. ಕೆಳಗಿನ ಅಂಶದೊಂದಿಗೆ *.env* ಎಂಬ ಫೈಲ್ ರಚಿಸಿ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. ಮೇಲಿನ ಗ್ರಂಥಾಲಯಗಳನ್ನು *requirements.txt* ಎನ್ನುತ್ತಿರುವ ಫೈಲಿನಲ್ಲಿ ಈ ರೀತಿ ಸೇರಿಸಿ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. ಬಳಿಕ, ವರ್ಚುವಲ್ ಎನ್ವಿರಾನ್‌ಮೆಂಟ್ ಅನ್ನು ರಚಿಸಿ ಮತ್ತು ಗ್ರಂಥಾಲಯಗಳನ್ನು ಸ್ಥಾಪಿಸಿರಿ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> ವಿಂಡೋಸ್‌ಗಾಗಿ, ನಿಮ್ಮ ವರ್ಚುವಲ್ ಪರಿಸರವನ್ನು ರಚಿಸಲು ಮತ್ತು ಸಕ್ರಿಯಗೊಳಿಸಲು ಕೆಳಗಿನ ನಿಯಮಗಳನ್ನು ಬಳಸಿ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ಎಂಬ ಫೈಲ್ನಲ್ಲಿ ಕೆಳಗಿನ ಕೋಡ್ ಅನ್ನು ಸೇರಿಸಿ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # import dotenv
    dotenv.load_dotenv()

    # OpenAI ವಸ್ತುವನ್ನು ರಚಿಸಿ
    client = openai.OpenAI()


    try:
        # ಚಿತ್ರ ರಚನೆ API ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸಿ
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಪ್ರಾಂಪ್ಟ್ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
            size='1024x1024',
            n=1
        )
        # ಸಂಗ್ರಹಿಸಿದ ಚಿತ್ರಕ್ಕೆ ಡೈರೆಕ್ಟರಿಯನ್ನು ಸೆಟ್ ಮಾಡಿ
        image_dir = os.path.join(os.curdir, 'images')

        # ಡೈರೆಕ್ಟರಿ ಲಭ್ಯವಿಲ್ಲದಿದ್ದಲ್ಲಿ, ಅದನ್ನು ರಚಿಸಿ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ಚಿತ್ರ ಮಾರ್ಗವನ್ನು ಪ್ರಾರಂಭಿಸಿ (ಖಾತ್ರಿ файл ಪ್ರಕಾರ png ಆಗಿರಬೇಕು)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # ರಚಿಸಲಾದ ಚಿತ್ರವನ್ನು ಪಡೆದುಕೊಳ್ಳಿ
        print(generation_response)

        image_url = generation_response.data[0].url # ಪ್ರತಿಕ್ರಿಯೆಯಿಂದ ಚಿತ್ರ URL ಅನ್ನು ತೆಗೆಯಿರಿ
        generated_image = requests.get(image_url).content  # ಚಿತ್ರವನ್ನು ಡೌನ್‌ಲೋಡ್ ಮಾಡಿ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ಡೀಫಾಲ್ಟ್ ಚಿತ್ರ ವೀಕ್ಷಣೆಯಲ್ಲಿ ಚಿತ್ರವನ್ನು ಪ್ರದರ್ಶಿಸಿ
        image = Image.open(image_path)
        image.show()

    # исключения ಹಿಡಿಕೊಳ್ಳಿ
    except openai.BadRequestError as err:
        print(err)

    ```

ನಾವು ಈ ಕೋಡ್ ಅನ್ನು ವಿವರಿಸೋಣ:

    ```python
    import openai
    import os

    import dotenv 
    ```

- ಅದನ್ನುಮೇಲೆ, ನಾವು ಆಬ್ಜೆಕ್ಟ್ ಅನ್ನು ರಚಿಸುತ್ತೇವೆ, ಇದು ನಿಮ್ಮ ``.env`` ನಿಂದ API ಕೀ ಅನ್ನು ಪಡೆಯುತ್ತದೆ.

    ```python
        # OpenAI ವಸ್ತುವನ್ನು ರಚಿಸಿ
        client = openai.OpenAI()
    ```

- ನಂತರ, ನಾವು ಚಿತ್ರವನ್ನು ರಚಿಸುತ್ತೇವೆ:

    ```python
    # ಚಿತ್ರ ರಚಿಸಲು ಚಿತ್ರ ಸೃಜನ API ಅನ್ನು ಬಳಸಿ
    generation_response = client.images.generate(
    ```

    ```python
    image = Image.open(image_path)
    ```

### ಚಿತ್ರ ರಚನೆ ಕುರಿತಾಗಿ ಹೆಚ್ಚಿನ ವಿವರಗಳು

ಚಿತ್ರವನ್ನು ರಚಿಸುವ ಕೋಡ್ ಏನಾಗಿದೆ ಎಂದು ವಿವರವಾಗಿ ನೋಡೋಣ:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ನಿಮ್ಮ ಸಂವೇದನೆ ಪಠ್ಯವನ್ನು ಇಲ್ಲಿ ನಮೂದಿಸಿ
        size='1024x1024',
        n=1
    )
```

- **prompt**, ಚಿತ್ರ ರಚಿಸಲು ಬಳಸಲಾಗುವ ಬರಹದ ಸೂಚನೆ ಆಗಿದೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, ನಾವು "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ಎಂಬ ಸೂಚನೆಯನ್ನು ಬಳಸುತ್ತಿದ್ದೇವೆ.
- **size**, ರಚನೆಗೊಳ್ಳುವ ಚಿತ್ರದ ಗಾತ್ರ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, ನಾವು 1024x1024 ಪಿಕ್ಸೆಲ್ಸ್ ಗಾತ್ರದ ಚಿತ್ರವನ್ನು ರಚಿಸುತ್ತಿದ್ದೇವೆ.
- **n**, ರಚಿಸಲ್ಪಡುವ ಚಿತ್ರಗಳ ಸಂಖ್ಯೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ, ನಾವು ಎರಡು ಚಿತ್ರಗಳನ್ನು ರಚಿಸುತ್ತಿದ್ದೇವೆ.

ಚಿತ್ರಗಳೊಂದಿಗೆ ನೀವು ಮಾಡಲು ಸಾಧ್ಯವಾದ ಇನ್ನಷ್ಟು ವಿಷಯಗಳಿವೆ, ಅವುಗಳನ್ನು ನಾವು ಮುಂದಿನ ವಿಭಾಗದಲ್ಲಿ ವಿವರಿಸುವೆವು.

## ಚಿತ್ರ ರಚನೆಗೆ ಹೆಚ್ಚುವರಿ ಸಾಮರ್ಥ್ಯಗಳು

ಈವರೆಗೂ ನೀವು ಕಂಡಂತಾಗಿದೆ ನಾವು ಕೆಲವು ಸಾಲುಗಳಲ್ಲಿ ಪೈಥಾನ್ ಬಳಸಿ ಚಿತ್ರವನ್ನು ರಚಿಸಬಹುದು. ಆದಾಗ್ಯೂ, ಚಿತ್ರಗಳೊಂದಿಗೆ ನೀವು ಇನ್ನಷ್ಟು ಮಾಡಲು ಬೇಕಾದವುಗಳಿವೆ.

ನೀವು ಕೆಳಗಿನವುಗಳನ್ನೂ ಮಾಡಬಹುದು:

- **ಎಡಿಟ್‌ಗಳನ್ನು ನಿರ್ವಹಿಸು**. ಇ موجود ಚಿತ್ರಕ್ಕೆ ಮಾಸ್ಕ್ ಮತ್ತು ಸೂಚನೆಯನ್ನು ನೀಡುವ ಮೂಲಕ ಚಿತ್ರವನ್ನು ಬದಲಾಯಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ಚಿತ್ರದ ಒಂದು ಭಾಗಕ್ಕೆ ಏನಾದರೂ ಸೇರಿಸಬಹುದು. ನಮ್ಮ ಬಣ್ಣಿಯ ಚಿತ್ರದ ಬಗ್ಗೆ વિચರಿಸಿ, ಬಣ್ಣಿಗೆ ಹ್ಯಾಟ್ ಸೇರಿಸಬಹುದು. ಅದನ್ನು ಮಾಡಲು ನೀವು ಚಿತ್ರ, ಬದಲಾವಣೆ ಬೇಕಾದ ಪ್ರದೇಶವನ್ನು ಗುರುತಿಸುವ ಮಾಸ್ಕ್ ಮತ್ತು ಮಾಡಬೇಕಾದ ಕೆಲಸವನ್ನು ಹೇಳುವ ಬರೆಹ ಸೂಚನೆಯನ್ನು ನೀಡಬೇಕು.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ಮೂಲ ಚಿತ್ರದಲ್ಲಿ ಮಾತ್ರ খರಗಿ ಇರುವೆಯಾ, ಆದರೆ ಕೊನೆಗೆ ಬಣ್ಣಿಯ ಮೇಲೆ ಹ್ಯಾಟ್ ಇರುವ ಚಿತ್ರ ಸಿಗುತ್ತದೆ.
    
- **ತಾರತಮ್ಯಗಳನ್ನು ರಚಿಸು**. ಆ ಭಾವನೆ ಎಂದರೆ ನೀವು ಇ ಇದ್ದ ಘಟನೆಯನ್ನು ತೆಗೆದು, ಅದರ ತಾರತಮ್ಯಗಳನ್ನು ರಚಿಸಬೇಕು ಎಂದು ಹೇಳಬಹುದು. ತಾರತಮ್ಯವನ್ನು ರಚಿಸಲು ನೀವು ಚಿತ್ರವನ್ನು ಮತ್ತು ಬರೆಹ ಸೂಚನೆಯನ್ನು ಕೊಡಿ ಮತ್ತು ಈ ಕೆಳಗಿನ ಕೋಡ್ ಅನ್ನು ಬಳಸಿರಿ:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
