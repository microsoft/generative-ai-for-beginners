## ಪರಿಚಯ

ಈ ಪಾಠವು ಒಳಗೊಂಡಿದೆ:
- ಕಾರ್ಯವಿಧಾನ ಕರೆಗೆ ಏನು ಮತ್ತು ಅದರ ಬಳಸುವ ಪ್ರಕರಣಗಳು
- OpenAI ಬಳಸಿ ಕಾರ್ಯವಿಧಾನ ಕರೆ ಹೇಗೆ ರಚಿಸುವುದು
- ಕಾರ್ಯವಿಧಾನ ಕರೆವನ್ನು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಹೇಗೆ ಸಂಯೋಜಿಸುವುದು

## ಅಧ್ಯಯನ ಗುರಿಗಳು

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ನಂತರ ನೀವು ಹೇಗೆ ಮಾಡಲು ಮತ್ತು ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬಲ್ಲಿರಿ:

- ಕಾರ್ಯವಿಧಾನ ಕರೆ ಬಳಸುವ ಉದ್ದೇಶ
- OpenAI ಸೇವೆ ಬಳಸಿ ಕಾರ್ಯವಿಧಾನ ಕರೆ ಸೆಟಪ್ ಮಾಡುವುದು
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಬಳಕೆ ಪ್ರಕರಣಕ್ಕಾಗಿ ಪರಿಣಾಮಕಾರಿ ಕಾರ್ಯವಿಧಾನ ಕರೆಗಳನ್ನು ವಿನ್ಯಾಸಗೊಳಿಸುವುದು


## ಕಾರ್ಯ ಕರೆಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು

ಈ ಪಾಠಕ್ಕಾಗಿ, ನಾವು ನಮ್ಮ ಶಿಕ್ಷಣ ಸ್ಟಾರ್ಟ್‌ಅಪ್‌ಗೆ ಒಂದು ವೈಶಿಷ್ಟ್ಯವನ್ನು ನಿರ್ಮಿಸಲು ಬಯಸುತ್ತೇವೆ, ಅದು ಬಳಕೆದಾರರು ತಾಂತ್ರಿಕ ಕೋರ್ಸುಗಳನ್ನು ಕಂಡುಹಿಡಿಯಲು ಚಾಟ್‌ಬಾಟ್ ಅನ್ನು ಬಳಸಲು ಅವಕಾಶ ನೀಡುತ್ತದೆ. ನಾವು ಅವರ ನಿಪುಣತೆ ಮಟ್ಟ, ಪ್ರಸ್ತುತ ಪಾತ್ರ ಮತ್ತು ಆಸಕ್ತಿಯ ತಂತ್ರಜ್ಞಾನಕ್ಕೆ ಅನುಗುಣವಾಗಿರುವ ಕೋರ್ಸುಗಳನ್ನು ಶಿಫಾರಸು ಮಾಡುವೆವು.

ಇದನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು ನಾವು ಸಂಯೋಜನೆ ಬಳಸುತ್ತೇವೆ:
 - ಬಳಕೆದಾರನಿಗಾಗಿ ಚಾಟ್ ಅನುಭವವನ್ನು ಸೃಷ್ಟಿಸಲು `OpenAI`
 - ಬಳಕೆದಾರನ ವಿನಂತಿಯ ಆಧಾರದ ಮೇಲೆ ಕೋರ್ಸುಗಳನ್ನು ಕಂಡುಹಿಡಿಯಲು ಸಹಾಯ ಮಾಡಲು `Microsoft Learn Catalog API`
 - ಬಳಕೆದಾರನ ಪ್ರಶ್ನೆಯನ್ನು ಪಡೆದು ಅದನ್ನು API ವಿನಂತಿ ಮಾಡಲು ಕಾರ್ಯಕ್ಕೆ ಕಳುಹಿಸಲು `ಕಾರ್ಯ ಕರೆ` ಬಳಸುವುದು

ಪ್ರಾರಂಭಿಸಲು, ನಾವು ಮೊದಲು ಏಕೆ ಕಾರ್ಯ ಕರೆ ಬಳಸಬೇಕೆಂದು ನೋಡೋಣ:


### ಏಕೆ ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್  

ನೀವು ಈ ಕೋರ್ಸ್‌ನಲ್ಲಿ ಯಾವುದೇ ಇತರ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ್ದು ಇದ್ದರೆ, ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗಳನ್ನು (LLMs) ಬಳಸುವ ಶಕ್ತಿಯನ್ನು ನೀವು ಅರ್ಥಮಾಡಿಕೊಂಡಿರಬಹುದು. ಆಶಿಸಿದ್ದಂತೆ ನೀವು ಅದೊಂದಿಗೆ ಕೆಲವು ಅವರ ಸೀಮಿತತೆಗಳನ್ನು ಕೂಡ ಕಾಣಬಹುದು.  

ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ OpenAI ಸೇವೆಯ ಒಂದು ವೈಶಿಷ್ಟ್ಯವಾಗಿದೆ, ಇದು ಕೆಳಗಿನ ಸವಾಲುಗಳನ್ನು ಮುಕ್ತಾಯಗೊಳಿಸಲು ವಿನ್ಯಾಸಗೊಳಿಸಲಾಗಿದೆ:  

ಅಸಮತೋಲನವಾದ ಪ್ರತಿಕ್ರಿಯೆ ಸ್ವರೂಪಣೆ:  
- ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಗೆ ಮುಂಚಿತವಾಗಿ, ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಯಿಂದ ಪ್ರತಿಕ್ರಿಯೆಗಳು ಅಸಂರಚಿತ ಮತ್ತು ಅಸಮತೋಲನವಾಗಿದ್ದವು. ಡೆವಲಪರ್‌ಗಳು ಪ್ರತಿಯೊಂದು ವಿಭಿನ್ನತೆಯನ್ನು ಹ್ಯಾಂಡಲ್ ಮಾಡಲು ಸಂಕೀರ್ಣ ಮಾನ್ಯತೆ ಕೋಡ್ ಬರೆಯಬೇಕಾಗಿದ್ದಿತು.  

ಹೊರಗಿನ ಡೇಟಾ ಏಕೀಕರಣದ ಮಿತಿಗಳು:  
- ಈ ವೈಶಿಷ್ಟ್ಯಕ್ಕೆ ಮೊದಲು, ಅಪ್ಲಿಕೇಶನ್‌ನ ಇತರ ಭಾಗಗಳ ಡೇಟಾವನ್ನು ಚಾಟ್ ಪ್ರ_CONTEXT ಹೊಂದಿಸಲು ಕಷ್ಟವಾಗುತ್ತಿತ್ತು.  

ಪ್ರತಿಕ್ರಿಯೆ ಸ್ವರೂಪುದನ್ನು ಮಾನ್ಯಗೊಳಿಸುವ ಮೂಲಕ ಮತ್ತು ಹೊರಗಿನ ಡೇಟಾ ಸಂಯೋಜನೆಯನ್ನು ಹಾಸುಹಂತದಿಂದ ಕೂಡಿಸುವ ಮೂಲಕ, ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಅಭಿವೃದ್ಧಿಯನ್ನು ಸರಳಗೊಳಿಸುತ್ತದೆ ಮತ್ತು ಹೆಚ್ಚುವರಿ ಮಾನ್ಯತೆ ಲಾಜಿಕ್ನ ಅವಶ್ಯಕತೆಯನ್ನು ಕಡಿಮೆ ಮಾಡುತ್ತದೆ.  

ಬಳಕೆದಾರರು "ಸ್ಟಾಕ್‌ಹೋಲ್ಮ್‌ನಲ್ಲಿ ಪ್ರಸ್ತುತ ಹವಾಮಾನವೇನು?" ಎಂದು ಉತ್ತರವನ್ನು ಪಡೆಯಲಾರರು. ಇದಕ್ಕೆ ಕಾರಣ, ಮಾದರಿಗಳು ತರಬೇತುಗೊಂಡ ಸಮಯದ ಡೇಟಾದಷ್ಟೇ ಸೀಮಿತವಾಗಿದ್ದವು.  

ಕೆಳಗಿನ ಉದಾಹರಣೆಯನ್ನು ನೋಡೋಣ, ಇದು ಈ ಸಮಸ್ಯೆಯನ್ನು ವಿವರಿಸುತ್ತದೆ:  

ನಾವು ವಿದ್ಯಾರ್ಥಿಗಳ ಡೇಟಾಬೇಸ್ ಸೃಷ್ಟಿಸಿ, ಅವರಿಗೆ ಸರಿಯಾದ ಕೋರ್ಸ್ ಅನ್ನು ಶಿಫಾರಸು ಮಾಡಬೇಕೆಂದರೆ ಹೇಗ olur ಎಂಬುದನ್ನು ನೋಡೋಣ. ಕೆಳಗಿನಲ್ಲಿ ವಿದ್ಯಾರ್ಥಿಗಳ ಎರಡು ವಿವರಣೆಗಳಿವೆ, ಅವು ಸಮಾನವಾದ ಡೇಟಾವನ್ನು ಹೊಂದಿವೆ.  


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ನಾವು ಈ ಮಾಹಿತಿಯನ್ನು ವಿಭಾಗಿಸಲು LLM ಗೆ ಕಳುಹಿಸಲು ಬಯಸುತ್ತೇವೆ. ಇದು ನಂತರ ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ API ಗೆ ಕಳುಹಿಸಲು ಅಥವಾ ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿ ಸಂಗ್ರಹಿಸಲು ಬಳಸಬಹುದು.

ನಾವು ಆಸಕ್ತಿ ಹೊಂದಿರುವ ಮಾಹಿತಿಯ ಬಗ್ಗೆ LLM ಗೆ ತಿಳಿಸುವ ಎರಡು ಇದೇ ರೀತಿ ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ಸೃಷ್ಟಿಸೋಣ:


ನಾವು ಇದನ್ನು ನಮ್ಮ ಉತ್ಪನ್ನಕ್ಕೆ ಮಹತ್ವದ್ದಾದ ಭಾಗಗಳನ್ನು ವಿಶ್ಲೇಷಿಸಲು LLM ಗೆ ಕಳುಹಿಸಲು ಬಯಸುತ್ತೇವೆ. ಆದ್ದರಿಂದ ನಾವು LLMಗೆ ಸೂಚಿಸಲು ಎರಡು ಅನುರೂಪವಾದ ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ರಚಿಸಬಹುದು: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ಈ ಎರಡು ಪ್ರಾಂಪ್ಟ್ಗಳನ್ನು ರಚಿಸಿದ ನಂತರ, ನಾವು ಅವುಗಳನ್ನು `client.responses.create` ಬಳಸಿ LLM ಗೆ ಕಳುಹಿಸುವೆವು. ನಾವು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು `input` 변수ನಲ್ಲಿ ಸಂಗ್ರಹಿಸಿ, ಪಾತ್ರವನ್ನು `user` ಗೆ ನಿಯೋಜಿಸುತ್ತೇವೆ. ಇದು ಬಳಕೆದಾರನಿಂದ ಚಾಟ್‌ಬಾಟ್ ಗೆ ಸಂದೇಶವನ್ನು ಬರೆಯುತ್ತಿರುವಂತೆ ನಕಲು ಮಾಡಲು. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"
```

ಈಗ ನಾವು ಇಬ್ಬರು ವಿನಂತಿಗಳನ್ನು LLM ಗೆ ಕಳುಹಿಸಿ ನಾವು ಪಡೆದ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪರಿಶೀಲಿಸಬಹುದು. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ಪ್ರಾಂಪ್ಟ್ಗಳು ಒಂದೇ ಆಗಿದ್ದರೂ ಹಾಗೂ ವಿವರಣೆಗಳು ಸಮಾನವಾಗಿದ್ದರೂ, ನಾವು `Grades` ಗುಣಲಕ್ಷಣದ ವಿಭಿನ್ನ ಸ್ವರೂಪಗಳನ್ನು ಪಡೆಯಬಹುದು.

ಮೇಲಿನ ಕೋಶವನ್ನು ಎಷ್ಟು ಬಾರಿ ನಡಸಿದರೂ, ಸ್ವರೂಪವು `3.7` ಅಥವಾ `3.7 GPA` ಆಗಿರಬಹುದು.

ಇದಕ್ಕೆ ಕಾರಣವೆಂದರೆ LLM ಬರವಣಿಗೆಯ ಪ್ರಾಂಪ್ಟ್ ರೂಪದಲ್ಲಿ ಅಸಂರಚಿತ ಡೇಟಾವನ್ನು ತೆಗೆದುಕೊಂಡು, ಅಸಂರಚಿತ ಡೇಟಾನೇ واپسಿಸುತ್ತದೆ. ನಾವು ಈ ಡೇಟಾವನ್ನು ಸಂಗ್ರಹಿಸುವಾಗ ಅಥವಾ ಬಳಕೆ ಮಾಡುವಾಗ ಏನು ನಿರೀಕ್ಷಿಸಬೇಕೆಂದು ತಿಳಿದುಕೊಳ್ಳಲು ಸಂರಚಿತ ಸ್ವರೂಪ ಇದೆಯೆಂಬುದು ಅಗತ್ಯ.

ಕಾರ್ಯಾತ್ಮಕ ಕರೆ ಮಾಡುವ ಮೂಲಕ, ನಾವು ಸಂರಚಿತ ಡೇಟಾವನ್ನು ಹಿಂದಿರುಗಿಸುವುದನ್ನು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಬಹುದು. ಕಾರ್ಯಾತ್ಮಕ ಕರೆ ಮಾಡುವಾಗ, LLM ನಿಜವಾಗಿಯಾಗಿ ಯಾವುದೇ ಕಾರ್ಯಗಳನ್ನು ಕರೆಯುವುದಿಲ್ಲ ಅಥವಾ ನಡೆಯುವುದಿಲ್ಲ. ಬದಲಿಗೆ, ಉತ್ತರಗಳಿಗೆ ಅನುಸರಿಸಲು LLM ಗೆ ಒಂದು ಸಂರಚನೆಯನ್ನು ನಾವು ರಚಿಸುತ್ತೇವೆ. ನಂತರ ನಾವು ಆ ಸಂರಚಿತ ಉತ್ತರಗಳನ್ನು ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ಗಳಲ್ಲಿ ಯಾವ ಕಾರ್ಯವನ್ನು ನಡೆಸಬೇಕು ಎಂದು ತಿಳಿದುಕೊಳ್ಳಲು ಬಳಸುತ್ತೇವೆ.

ಚಾಟ್‌ಬಾಟ್‌ಗಳು ಬಳಕೆದಾರರ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರಗಳನ್ನು ನೀಡುವಲ್ಲಿ ಅದ್ಭುತವಾಗಿವೆ. ಫಂಕ್ಷನ್ ಕರೆಗಳನ್ನು ಬಳಸಿ, ಚಾಟ್‌ಬಾಟ್‌ಗಳು ಬಳಕೆದಾರರ ಸಂದೇಶಗಳನ್ನು ಬಳಸಿಕೊಂಡು ನಿರ್ದಿಷ್ಟ ಕಾರ್ಯಗಳನ್ನು ಪೂರ್ಣಗೊಳಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ಒಂದು ವಿದ್ಯಾರ್ಥಿ ಚಾಟ್‌ಬಾಟ್ ಅನ್ನು ಕೇಳಬಹುದು "ನನ್ನ ಶಿಕ್ಷಕರಿಗೆ ಈ ವಿಷಯದಲ್ಲಿ ಮತ್ತಷ್ಟು ಸಹಾಯ ಬೇಕು ಎಂದು ಇಮೇಲ್ ಕಳುಹಿಸಿ". ಇದರಿಂದ `send_email(to: string, body: string)` ಎಂಬ ಫಂಕ್ಷನ್ ಕರೆ ಮಾಡಬಹುದು


**API ಅಥವಾ ಡೇಟಾಬೇಸ್ ಪ್ರಶ್ನೆಗಳನ್ನು ರಚಿಸುವುದು**
ಬಳಕೆದಾರರು ಪ್ರಕೃತಿಭಾಷೆ ಬಳಸಿ ಮಾಹಿತಿ ಕಂಡುಹಿಡಿಯೋಬಹುದು, ಅದು ಸೂಕ್ತವಾದ ಕ್ಯೂರಿ ಅಥವಾ API ವಿನಂತಿಯಾಗಿ ಪರಿವರ್ತಿಸಲಾಗುತ್ತದೆ. ಉದಾಹರಣೆಗೆ, "ಕೊನೆಯ ನಿಯೋಜನೆಯನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ವಿದ್ಯಾರ್ಥಿಗಳು ಯಾರು?" ಎಂದು ಕೇಳುವ ಶಿಕ್ಷಕರು `get_completed(student_name: string, assignment: int, current_status: string)` ಎಂಬ ಫಂಕ್ಷನ್ ಅನ್ನು ಕರೆ ಮಾಡಬಹುದು


**ಸಂರಚಿತ ಡೇಟಾ ಸೃಷ್ಟಿಸುವುದು**
ಬಳಕೆದಾರರು ಒಂದು ಪಠ್ಯ ಭಾಗ ಅಥವಾ CSV ತೆಗೆದುಕೊಂಡು ಅದರ‌ನಿಂದ ಮಹತ್ವದ ಮಾಹಿತಿಯನ್ನು LLM ಅನ್ನು ಬಳಸಿಕೊಂಡು ಪಡೆಯಬಹುದು. ಉದಾಹರಣೆಗೆ, ಒಬ್ಬ ವಿದ್ಯಾರ್ಥಿ ಶಾಂತಿ ಒಪ್ಪಂದಗಳ ಕುರಿತು ವಿಕಿಪೀಡಿಯ ಲೇಖನವನ್ನು AI ಫ್ಲಾಷ್ ಕಾರ್ಡ್ ಗಳನ್ನು ಸೃಷ್ಟಿಸಲು ಪರಿವರ್ತಿಸಬಹುದು. ಇದನ್ನು `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` ಎಂಬ ಫಂಕ್ಷನ್ ಬಳಸಿ ಮಾಡಬಹುದು

## 3. ಫಂಕ್ಷನ್ ಕಾಲ್‌ಗಳನ್ನು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಏಕತ್ರೀಕರಿಸುವುದು.


LLM ನಿಂದ ಸ್ವರೂಪಿತ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪರೀಕ್ಷಿಸಿದ ನಂತರ, ಈಗ ನಾವು ಇದನ್ನು ಒಂದು ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಏಕತ್ರೀಕರಿಸಬಹುದು.

### ಫ್ಲೋವನ್ನು ನಿರ್ವಹಿಸುವುದು

ಇದನ್ನು ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಏಕತ್ರೀಕರಿಸಲು, ಮುಂದಿನ ಕ್ರಮಗಳನ್ನು ತೆಗೆದುಕೊಳ್ಳೋಣ:

ಮೊದಲಿಗೆ, OpenAI ಸೇವೆಗಳಿಗೆ ಕರೆ ಮಾಡಿ ಮತ್ತು ಸಂದೇಶವನ್ನು `response_message` ಎಂಬ ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸೋಣ.


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ಈಗ ನಾವು Microsoft Learn API ಅನ್ನು ಕರೆಮಾಡಿ ಕೋರ್ಸ್‌ಗಳ ಪಟ್ಟಿ ಪಡೆಯಲಿರುವ ಫಂಕ್ಷನ್ ಅನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಲು ಹೋಗುತ್ತಿದ್ದೇವೆ: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ಉತ್ತಮ ಅಭ್ಯಾಸವಾಗಿ, ನಾವು ನಂತರ ಮಾದರಿ ಯಾವುದೇ ಫಂಕ್ಷನ್ ಕರೆದಕೊಳ್ಳಲು ಇಚ್ಛಿಸುತ್ತದೆ ಎಂದು ನೋಡುತ್ತೇವೆ. ನಂತರ, ನಾವು ಲಭ್ಯವಿರುವ ಫಂಕ್ಷನ್‌ಗಳಲ್ಲಿ ಒಂದನ್ನು ರಚಿಸಿ ಅದನ್ನು ಕರೆಮಾಡಲಾದ ಫಂಕ್ಷನ್‌ಗೆ ಹೊಂದಿಸುತ್ತೇವೆ. 
ನಂತರ ನಾವು ಫಂಕ್ಷನ್‌ನ ಆರ್ಗ್ಯುಮೆಂಟ್ಗಳನ್ನು ತೆಗೆದುಕೊಂಡು ಅವುಗಳನ್ನು LLM‌ನ ಆರ್ಗ್ಯುಮೆಂಟ್ಗಳಿಗೆ ನಕ್ಷೆ ಮಾಡುತ್ತೇವೆ.

ಕೊನೆಗೆ, ನಾವು ಫಂಕ್ಷನ್ ಕಾಲ್ ಸಂದೇಶ ಮತ್ತು `search_courses` ಸಂದೇಶದಿಂದ ಮರಳಿಸಲಾಗಿರುವ ಮೌಲ್ಯಗಳನ್ನು ಜೋಡಿಸುತ್ತೇವೆ. ಇದರಿಂದ LLM ಗೆ ಎಲ್ಲಾ ಅಗತ್ಯ ಮಾಹಿತಿಗಳು ಸಿಗುತ್ತವೆ
ಬಳಕೆದಾರನಿಗೆ ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿಕೊಂಡು ಪ್ರತಿಕ್ರಿಯೆ ನೀಡಲು. 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```
