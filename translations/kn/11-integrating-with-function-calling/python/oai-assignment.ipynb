## ಪರಿಚಯ

ಈ ಪಾಠದಲ್ಲಿ ಈ ವಿಷಯಗಳನ್ನು ನವೀಕರಿಸಲಾಗುತ್ತದೆ:
- ಫಂಕ್ಷನ್ ಕರೆ ಮಾಡುವುದೇನು ಮತ್ತು ಅದರ ಬಳಕೆ ಪ್ರಕರಣಗಳು
- OpenAI ಬಳಸಿ ಫಂಕ್ಷನ್ ಕರೆಗಳನ್ನು ಹೇಗೆ ರಚಿಸುವುದು
- ಫಂಕ್ಷನ್ ಕರೆಗಳನ್ನು ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ಹೇಗೆ ಏಕರೂಪಗೊಳಿಸುವುದು

## ಶಿಕ್ಷಣ ಗುರಿಗಳು

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ನಂತರ ನೀವು ತಿಳಿದುಕೊಳ್ಳುವಿರಿ ಮತ್ತು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವಿರಿ:

- ಫಂಕ್ಷನ್ ಕರೆ ಬಳಕೆ ಉದ್ದೇಶವೇನು
- OpenAI ಸೇವೆ ಬಳಸಿ ಫಂಕ್ಷನ್ ಕರೆ ವ್ಯವಸ್ಥೆಯನ್ನು ಹೊಂದಿಸುವುದು
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ಗಾಗಿ ಪರಿಣಾಮಕಾರಿ ಫಂಕ್ಷನ್ ಕರೆಗಳನ್ನು ವಿನ್ಯಾಸಗೊಳಿಸುವುದು


## ಫಂಕ್ಷನ್ ಕರೆಗಳ ಅರ್ಥೈಸಿಕೆ

ಈ ಪಾಠಕ್ಕಾಗಿ, ನಮ್ಮ ಶಿಕ್ಷಣ ಸ್ಟಾರ್ಟಪ್‌ಗೆ ಒಂದು ಐಶಾರಾಮಿ ಗುಣವನ್ನು ನಿರ್ಮಿಸಬೇಕು, ಅದು ಬಳಕೆದಾರರಿಗೆ ಟೆಕ್ನಿಕಲ್ ಕೋರ್ಸುಗಳನ್ನು ಹುಡುಕಲು ಚಾಟ್‌ಬಾಟ್ ಬಳಸಲು ಅವಕಾಶ ನೀಡುತ್ತದೆ. ನಾವು ಅವರ ಕುಶಲತಾ ಮಟ್ಟ, ಪ್ರಸ್ತುತ ಪಾತ್ರ ಮತ್ತು ಆಸಕ್ತಿ ಹೊಂದಿರುವ ತಂತ್ರಜ್ಞಾನಕ್ಕೆ сәйкес ಕೋರ್ಸುಗಳನ್ನು ಶಿಫಾರಸು ಮಾಡುತ್ತೇವೆ.

ಇದನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು ನಾವು ಸಂಯೋಜನೆಯು ಕೆಳಗಿನಂತಿರುತ್ತದೆ:
 - ಬಳಕೆದಾರನಿಗೆ ಚಾಟ್ ಅನುಭವವನ್ನು ಸೃಷ್ಟಿಸಲು `OpenAI`
 - ಬಳಕೆದಾರನ ವಿನಂತಿಯ ಆಧಾರದಲ್ಲಿ ಕೋರ್ಸುಗಳನ್ನು ಹುಡುಕಲು ಸಹಾಯ ಮಾಡಲು `Microsoft Learn Catalog API`
 - ಬಳಕೆದಾರನ ಪ್ರಶ್ನೆಯನ್ನು ತೆಗೆದುಕೊಂಡು API ವಿನಂತಿಯನ್ನು ಮಾಡಲು ಫಂಕ್ಷನ್‌ಗೆ ಕಳುಹಿಸಲು `Function Calling`

ಪ್ರಾರಂಭಿಸಲು, ನಾವು ಮೊದಲಿಗೆ ಏಕೆ ಫಂಕ್ಷನ್ ಕರೆಗಳನ್ನು ಬಳಸಬೇಕೆಂದು ನೋಡೋಣ:


### ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆಯ ಅಗತ್ಯವೇನು

ನೀವು ಈ ಕೋರ್ಸ್‌ನ ಯಾವುದೇ ಇನ್ನೊಂದು ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ್ದರೆ, ನೀವು ಬಹುಶಃ ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗಳನ್ನು (LLMs) ಉಪಯೋಗಿಸುವ ಶಕ್ತಿಯನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುತ್ತೀರಿ. ಆಶಾಭಂಗವಾಗಿ, ನೀವು ಅವುಗಳ ಕೆಲವು ಮಿತಿಗಳನ್ನು ಸಹ ನೋಡಬಹುದು.

ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆವು OpenAI ಸೇವೆಯ一个 ವೈಶಿಷ್ಟ್ಯವಾಗಿದ್ದು, ಕೆಳಗಿನ ಸವಾಲುಗಳನ್ನು ಪರಿಹರಿಸಲು రూపొందಿಸಲಾಗಿದೆ:

ಅಸಮಾನ ಪ್ರತಿಕ್ರಿಯೆ ವಿನ್ಯಾಸ:
- ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆಗೆ ಮೊದಲೆಯೇ, ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಯಿಂದ ಪ್ರತಿಕ್ರಿಯೆಗಳು ಅಸಂರಚಿತವಾಗಿದ್ದವು ಮತ್ತು ಅಸಮಾನವಾಗಿದ್ದವು. ಡೆವಲಪರ್ಗಳು ಪ್ರತಿಯೊಂದು ಬದಲಾವಣೆಯನ್ನು ನಿರ್ವಹಿಸಲು ಸಂಕೀರ್ಣ ಮಾನ್ಯತೆ ಕೋಡ್ ಬರೆಯಬೇಕಾಗಿತ್ತು.

ಹೊರಗಿನ ಡೇಟಾ ಜೊತೆಗೆ ಸೀಮಿತ ಏಕೀಕರಣ:
- ಈ ವೈಶಿಷ್ಟ್ಯಕ್ಕೆ ಮೊದಲು, ಚಾಟ್ ಸಂದರ್ಭದಲ್ಲಿ ಅನ್ವಯದ ಇತರ ಭಾಗಗಳಿಂದ ಡೇಟಾವನ್ನು ಸೇರಿಸುವುದು ಕಷ್ಟವಾಗಿತ್ತು.

ಪ್ರತಿಕ್ರಿಯೆಗಳ ವಿನ್ಯಾಸಗಳನ್ನು ಮಾನಕಗೊಳಿಸುವ ಮೂಲಕ ಮತ್ತು ಹೊರಗಿನ ಡೇಟಾ ಜೊತೆ ಸರ್‍ಲಭ ಏಕೀಕರಣವನ್ನು ಸಾದ್ಯಮಾಡುತ್ತ, ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆ ಅಭಿವೃದ್ಧಿಯನ್ನು ಸರಳಗೊಳಿಸುತ್ತದೆ ಮತ್ತು ಹೆಚ್ಚುವರಿ ಮಾನ್ಯತೆ ಲಾಜಿಕ್ನ ಅವಶ್ಯಕತೆಯನ್ನು ಕಡಿಮೆ ಮಾಡುತ್ತದೆ.

ಬಳಕೆದಾರರು "ಸ್ಟಾಕ್‌ಹೊಲ್ಮ್ನಲ್ಲಿ ಪ್ರಸ್ತುತ ಹವಾಮಾನ ಯಾವುದು?" ಎಂಬಂತಹ ಉತ್ತರಗಳನ್ನು ಪಡೆಯಲಾಗುತ್ತದೆಯೆಂದು ಕೆಳಗೆ ನೋಡಬಹುದು. ಇದಕ್ಕೆ ಕಾರಣ, ಮಾದರಿಗಳು ತರಬೇತಿ ಪಡೆದ ಸಮಯದ ಡೇಟಾ ಮಾತ್ರಕ್ಕೆ ಸೀಮಿತವಾಗಿವೆ.

ಈ ಸಮಸ್ಯೆಯನ್ನು ವಿವರಿಸುವ ಕೆಳಗಿನ ಉದಾಹರಣೆಯನ್ನು ನೋಡೋಣ:

ವಿದ್ಯಾರ್ಥಿಗಳ ಡೇಟಾಬೇಸ್ ರಚಿಸುವುದನ್ನು ನಾವು ಬಯಸುವುದಾಗಿ ಹೇಳಿ, ಅವರಿಗೆ ಸರಿಯಾದ ಕೋರ್ಸ್ ಸೂಚಿಸಲು. ಕೆಳಗಿನ ವಿವರಣೆಗಳಲ್ಲಿ ಎರಡು ವಿದ್ಯಾರ್ಥಿಗಳ ಡೇಟಾ ಬಹಳ ಸಮಾನವಾಗಿದೆ.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ನಾವು ಇದನ್ನು ಡೇಟಾವನ್ನು ಪಾರ್ಸ್ ಮಾಡಲು ಒಂದು LLM ಗೆ ಕಳುಹಿಸಲು ಬಯಸುತ್ತೇವೆ. ಇದನ್ನು ಮುಂದೆಯಲ್ಲಿ ನಮಗೆ ನಮ್ಮ ಅನ್ವಯದಲ್ಲಿ API ಗೆ ಕಳುಹಿಸಲು ಅಥವಾ ಡೇಟಾಬೇಸ್ ನಲ್ಲಿ ಸಂಗ್ರಹಿಸಲು ಬಳಸಬಹುದು.

ನಾವು ಎರಡು ಒಂದೇ ರೀತಿಯ ಸೂಚನೆಗಳನ್ನು ಸೃಜಿಸೋಣ, ನಾವು ಯಾವ ಮಾಹಿತಿಯಲ್ಲಿ ಆಸಕ್ತಿ ಹೊಂದಿದ್ದೇವೆ ಎನ್ನುವುದನ್ನು LLM ಗೆ ನಿರ್ದೇಶಿಸುವಂತೆ:


ನಾವು ಇದನ್ನು ನಮ್ಮ ಉತ್ಪನ್ನಕ್ಕೆ ಮಹತ್ವವಾದ ಭಾಗಗಳನ್ನು ವಿಶ್ಲೇಷಿಸಲು LLM ಗೆ ಕಳುಹಿಸಲು ಬಯಸುತ್ತೇವೆ. ಆದ್ದರಿಂದ ನಾವು LLM ಅನ್ನು ಸೂಚಿಸಲು ಎರಡು ಸಮಾನ ಪ್ರಾಂಪ್ಟ್ಗಳನ್ನು ರಚಿಸಬಹುದು: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ಈ ಎರಡು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ರಚಿಸಿದ ನಂತರ, ನಾವು ಅವುಗಳನ್ನು `client.responses.create` ಬಳಸಿ LLM ಗೆ ಕಳುಹಿಸುತ್ತೇವೆ. ನಾವು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು `input` ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸುತ್ತೇವೆ ಮತ್ತು ಪಾತ್ರವನ್ನು `user` ಗೆ ನಿಯೋಜಿಸುತ್ತೇವೆ. ಇದು ಚಾಟ್‌ಬಾಟ್‌ಗೆ ಬಳಕೆದಾರರಿಂದ ಸಂದೇಶ ಬರೆಯುವಂತೆ ಅನುಕರಿಸಲು עבור.



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"
```

ಈಗ ನಾವು ಎರಡೂ ವಿನಂತಿಗಳನ್ನು LLM ಗೆ ಕಳುಹಿಸಿ ನಾವು ಪಡೆದ ಉತ್ತರವನ್ನು ಪರಿಶೀಲಿಸಬಹುದು.


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ಪ್ರಾಂಪ್ಟ್ಗಳು ಒಂದೇ ಇದ್ದರೂ ಮತ್ತು ವಿವರಣೆಗಳು ಸಮಾನವಾಗಿದ್ದರೂ, ನಾವು `Grades` ಗುಣಲಕ್ಷಣದ ವಿಭಿನ್ನ ಸ್ವರೂಪಗಳನ್ನು ಪಡೆದುಕೊಳ್ಳಬಹುದು. 

ಮೇಲಿನ ಸೆಲ್ ಅನ್ನು ಬಹುಮಾನವಾಗಿ ರನ್ ಮಾಡಿದರೆ, ಸ್ವರೂಪ `3.7` ಅಥವಾ `3.7 GPA` ಆಗಿರಬಹುದು. 

ಇದಕ್ಕೆ ಕಾರಣವೆಂದರೆ LLM ಬರೆಯಲಾದ ಪ್ರಾಂಪ್ಟ್ ರೂಪದಲ್ಲಿ ಸಂರಚಿತವಲ್ಲದ ಡೇಟಾವನ್ನು ತೆಗೆದುಕೊಂಡು, ಅದನ್ನೇ ಸಂರಚಿತವಲ್ಲದ ಡೇಟಾವಾಗಿ ಹಿಂತಿರುಗಿಸುತ್ತದೆ. ನಾವು ಸಂರಚಿತ ಸ್ವರೂಪವನ್ನು ಹೊಂದಿರಬೇಕು ಎಂದು ತಿಳಿದಿರಬೇಕು, ನಾವು ಈ ಡೇಟಾವನ್ನು ಸಂಗ್ರಹಿಸುವಾಗ ಅಥವಾ ಬಳಸುವಾಗ ಏನು ನಿರೀಕ್ಷಿಸಬಹುದು ಎಂದು ತಿಳಿದುಕೊಳ್ಳಲು.

ಫಂಕ್ಷನಲ್ ಕರೆಯುವಿಕೆಯ ಬಳಕೆಯಿಂದ, ನಾವು ಸಂರಚಿತ ಡೇಟಾವನ್ನು ಹಿಂತಿರುಗಿಸುವುದನ್ನು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಬಹುದು. ಫಂಕ್ಷನ್ ಕರೆಯುವಿಕೆಯನ್ನು ಬಳಸುವುದರಲ್ಲಿ, LLM ನಿಜವಾಗಿಯೂ ಯಾವುದೇ ಫಂಕ್ಷನ್‌ಗಳನ್ನು ಕರೆಯುವುದಿಲ್ಲ ಅಥವಾ ಚಾಲನಾ ಮಾಡದು. ಬದಲಾಗಿ, ನಾವು LLM ಅನುಸರಿಸುವ ಪ್ರತಿಕ್ರಿಯೆಗಳಿಗಾಗಿ ಒಂದು ಸಂರಚನೆಯನ್ನು ಸೃಷ್ಟಿಸುತ್ತೇವೆ. ನಂತರ ಆ ಸಂರಚಿತ ಪ್ರತಿಕ್ರಿಯೆಗಳನ್ನು ನಮ್ಮ ಅನ್ವಯಿಕೆಗಳಲ್ಲಿ ಯಾವ ಫಂಕ್ಷನ್ ಓಡಿಸುವುದು ಎಂದು ತಿಳಿದುಕೊಳ್ಳಲು ಬಳಸುತ್ತೇವೆ.

ಚಾಟ್‌ಬಾಟ್‌ಗಳು ಬಳಕೆದಾರರಿಂದ ಬಂದ ಪ್ರಶ್ನೆಗಳಿಗೆ ಉತ್ತರ ನೀಡಲು ಉತ್ತಮವಾಗಿವೆ. ಕಾರ್ಯಾಚರಣೆ ಕರೆಗಳನ್ನು ಬಳಸುವ ಮೂಲಕ, ಚಾಟ್‌ಬಾಟ್‌ಗಳು ಬಳಕೆದಾರರ ಸಂದೇಶಗಳನ್ನು ಕೆಲ ಕಾರ್ಯಗಳನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು ಬಳಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ವಿದ್ಯಾರ್ಥಿ "ನನಗೆ ಈ ವಿಷಯದಲ್ಲಿ ಹೆಚ್ಚು ಸಹಾಯ ಬೇಕೆಂದು ನನ್ನ ಶಿಕ್ಷಕರಿಗೆ ಇಮೇಲ್ ಕಳುಹಿಸಿ" ಎಂದು ಚಾಟ್‌ಬಾಟ್‌ಗೆ ಕೇಳಬಹುದು. ಇದು `send_email(to: string, body: string)` ಎಂಬ ಕಾರ್ಯಾಚರಣೆಯನ್ನು ಕರೆದಿಕೊಳ್ಳಬಹುದು  


**API ಅಥವಾ ಡೇಟಾಬೇಸ್ ಪ್ರಶ್ನೆಗಳ ರಚನೆ**  
ಬಳಕೆದಾರರು ಸಹಜ ಭಾಷೆಯನ್ನು ಬಳಸಿಕೊಂಡು ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಬಹುದು, ಇದು ರೂಪುಗೊಂಡ ಪ್ರಶ್ನೆ ಅಥವಾ API ವಿನಂತಿಗೆ ಪರಿವರ್ತನೆಯಾಗುತ್ತದೆ. ಉದಾಹಣೆಗೆ, "ಇತ್ತೀಚಿನ ಕಾರ್ಯವನ್ನು ಪೂರ್ಣಗೊಂಡ ವಿದ್ಯಾರ್ಥಿಗಳು ಯಾರಾಗಿದ್ದಾರೆ" ಎಂದು ಕೇಳುವ ಶಿಕ್ಷಕನಿಗೆ `get_completed(student_name: string, assignment: int, current_status: string)` ಎಂಬ ಕಾರ್ಯಾಚರಣೆ ಕರೆ ಮಾಡಬಹುದು  


**ಸಂರಚಿತ ಡೇಟಾ ಸೃಷ್ಟಿಸುವುದು**  
ಬಳಕೆದಾರರು ಒಂದು ಪಠ್ಯ ಬ್ಲಾಕ್ ಅಥವಾ CSV ತೆಗೆದುಕೊಂಡು ಅದರಿಂದ ಪ್ರಮುಖ ಮಾಹಿತಿಯನ್ನು LLM ಬಳಸಿ ಹೊರತೆಗೆಯಬಹುದು. ಉದಾಹರಣೆಗಾಗಿ, ವಿದ್ಯಾರ್ಥಿ ಶಾಂತಿ ಒಪ್ಪಂದಗಳ ಬಗ್ಗೆ ವಿಕಿಪೀಡಿಯಾ ಲೇಖನವನ್ನು AI ಫ್ಲಾಷ್‌ಕಾರ್ಡ್‌ಗಳಾಗಿ ಪರಿವರ್ತಿಸಬಹುದು. ಇದನ್ನು `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` ಎಂಬ ಕಾರ್ಯಾಚರಣೆ ಬಳಸಿ ಮಾಡಬಹುದು  


## 2. ನಿಮ್ಮ ಮೊದಲ ಫಂಕ್ಷನ್ ಕರೆ ರಚಿಸುವುದು 

ಫಂಕ್ಷನ್ ಕರೆಯನ್ನು ರಚಿಸುವ ಪ್ರಕ್ರಿಯೆಯಲ್ಲಿ 3 ಮುಖ್ಯ ಹಂತಗಳಿವೆ: 
1. ನಿಮ್ಮ ಫಂಕ್ಷನ್‌ಗಳ ಪಟ್ಟಿ ಮತ್ತು ಬಳಕೆದಾರ ಸಂದೇಶವನ್ನು ಬಳಸಿ ಚಾಟ್ ಸಮಾಪ್ತಿಗಳ API ಅನ್ನು ಕರೆ ಮಾಡುವುದು 
2. ಕ್ರಿಯೆಯನ್ನು ಮಾಡುವ/ನಿರ್ವಹಿಸುವುದಕ್ಕೆ ಮಾದರಿಯ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಓದಿಕೊಳ್ಳುವುದು ಹೀಗೆ ಫಂಕ್ಷನ್ ಅಥವಾ API ಕರೆ ನಡೆಸುವುದು 
3. ಬಳಕೆದಾರನಿಗೆ ಪ್ರತಿಕ್ರಿಯೆ ರಚಿಸಲು ಆ ಮಾಹಿತಿಯನ್ನು ಬಳಸಲು ನಿಮ್ಮ ಫಂಕ್ಷನ್‌ನ ಪ್ರತಿಕ್ರಿಯೆಯಿಂದ ಚಾಟ್ ಸಮಾಪ್ತಿಗಳ API ಅನ್ನು ಮತ್ತೊಂದು ಬಾರಿ ಕರೆ ಮಾಡುವುದು.

### ಕಾರ್ಯಾನ್ವಯ ಕರೆಮಾಡುವ ಅಂಶಗಳು 

#### ಬಳಕೆದಾರರ ಇನ್ಪುಟ್ 

ಮೊದಲ ಹಂತವು ಬಳಕೆದಾರ ಸಂದೇಶವನ್ನು ರಚಿಸುವುದು. ಇದನ್ನು ಪಠ್ಯ ಇನ್ಪುಟ್ ಮೌಲ್ಯ ತೆಗೆದು ಸಸ್ಸ್ವರೂಪವಾಗಿ ನಿಯೋಜಿಸಬಹುದು ಅಥವಾ ನೀವು ಇಲ್ಲಿ ಮೌಲ್ಯವನ್ನು ನಿಯೋಜಿಸಬಹುದು. ನೀವು ಮೊದಲಬಾರಿಗೆ Chat Completions API ಜೊತೆ ಕೆಲಸ ಮಾಡುತ್ತಿದ್ದರೆ, ಸಂದೇಶದ `role` ಮತ್ತು `content` ಅನ್ನು ನಿರ್ಧರಿಸಬೇಕಾಗುತ್ತದೆ. 

`role` ಇವುಗಳಾಗಬಹುದು: `system` (ನಿಯಮಗಳನ್ನು ರಚಿಸುವುದು), `assistant` (ಮಾದರಿ) ಅಥವಾ `user` (ಅಂತಿಮ ಬಳಕೆದಾರ). ಕಾರ್ಯಾನ್ವಯ ಕರೆಗಾಗಿ, ನಾವು ಇದನ್ನು `user` ಮತ್ತು ಉದಾಹರಣೆಯ ಪ್ರಶ್ನೆಯಾಗಿ ನಿಯೋಜಿಸುತ್ತೇವೆ. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ಕಾರ್ಯಗಳನ್ನು ರಚಿಸುವುದು. 

ಮುಂದಿನದಾಗಿ ನಾವು ಕಾರ್ಯವನ್ನು ಮತ್ತು ಆ ಕಾರ್ಯದ ಪ್ಯಾರಾಮೀಟರ್ಗಳನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಲಿದ್ದೇವೆ. ನಾವು ಇಲ್ಲಿ `search_courses` ಎಂದು ಕರೆದಿರುವ ಒಂದು ಕಾರ್ಯವನ್ನು ಮಾತ್ರ ಬಳಸುತ್ತೇವೆ ಆದರೆ ನೀವು ಹಲವಾರು ಕಾರ್ಯಗಳನ್ನು ರಚಿಸಬಹುದು.

**ಮುಖ್ಯ** : ಕಾರ್ಯಗಳನ್ನು LLM‌ಗೆ ವ್ಯವಸ್ಥೆಯ ಸಂದೇಶದಲ್ಲಿ ಸೇರಿಸಲಾಗುತ್ತದೆ ಮತ್ತು ನೀವು ಬಳಸಬಹುದಾದ ಲಭ್ಯವಿರುವ ಟೋಕನ್‌ಗಳ ಪ್ರಮಾಣದಲ್ಲಿಯೂ ಸೇರಿಸಲಾಗುತ್ತದೆ. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**ವ್ಯಾಖ್ಯಾನಗಳು** 

ಫಂಕ್ಷನ್ ವ್ಯಾಖ್ಯಾನ ರಚನೆಯಲ್ಲಿ ಹಲವಾರು ಮಟ್ಟಗಳಿವೆ, ಪ್ರತಿ ಒಂದೂ ತನ್ನ ಸ್ವಂತ ಗುಣಲಕ್ಷಣಗಳನ್ನು ಹೊಂದಿದೆ. ಇಲ್ಲಿದೆ ಆ ನೆಸ್ಟೆಡ್ ರಚನೆಯ ವಿಭಜನೆ:

**ಟಾಪ್ ಲೆವೆಲ್ ಫಂಕ್ಷನ್ ಗುಣಲಕ್ಷಣಗಳು:**

`name` - ನಾವು ಕರೆ ಮಾಡಲು ಇಚ್ಛಿಸುವ ಫಂಕ್ಷನ್ ನ ಹೆಸರು. 

`description` - ಈ ಫಂಕ್ಷನ್ ಹೇಗೆ ಕೆಲಸ ಮಾಡುತ್ತದೆ ಎಂಬ ವಿವರಣೆ. ಇಲ್ಲಿ ಸ್ಪಷ್ಟವಾಗಿರುವುದು ಮತ್ತು ವಿಶಿಷ್ಟವಾಗಿರುವುದು ಮುಖ್ಯ 

`parameters` - ಮಾದರಿಯಿಂದ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಉತ್ಪಾದಿಸಲು ನೀವು ಬಯಸುವ ಮೌಲ್ಯಗಳ ಮತ್ತು ಸ್ವರೂಪಗಳ ಪಟ್ಟಿ 

**ಪ್ಯಾರಾಮೀಟರ್ ಆಬ್ಜೆಕ್ಟ್ ಗುಣಲಕ್ಷಣಗಳು:**

`type` - ಪ್ಯಾರಾಮೀಟರ್ ಆಬ್ಜೆಕ್ಟ್ ಡೇಟಾ ಪ್ರಕಾರ (ಸಾಧಾರಣವಾಗಿ "object")

`properties` - ಮಾದರಿ ತನ್ನ ಪ್ರತಿಕ್ರಿಯೆಗೆ ಬಳಸುವ ನಿರ್ದಿಷ್ಟ ಮೌಲ್ಯಗಳ ಪಟ್ಟಿ 

**ವೈಯಕ್ತಿಕ ಪ್ಯಾರಾಮೀಟರ್ ಗುಣಲಕ್ಷಣಗಳು:**

`name` - ಪ್ರಾಬ್ಯಾಂಧಾತ್ಮಕವಾಗಿ ಪ್ರಾಪರ್ಟಿ ಕೀ ಮೂಲಕ ವ್ಯಾಖ್ಯಾನಿಸಲಾಗಿದೆ (ಉದಾ: "role", "product", "level")

`type` - ಈ ನಿರ್ದಿಷ್ಟ ಪ್ಯಾರಾಮೀಟರ್‌ನ ಡೇಟಾ ಪ್ರಕಾರ (ಉದಾ: "string", "number", "boolean") 

`description` - ನಿರ್ದಿಷ್ಟ ಪ್ಯಾರಾಮೀಟರ್ ನ ವಿವರಣೆ 

**ಐಚ್ಛಿಕ ಗುಣಲಕ್ಷಣಗಳು:**

`required` - ಫಂಕ್ಷನ್ ಕರೆ ಪೂರ್ಣಗೊಳ್ಳಲು ಅಗತ್ಯವಿರುವ ಪ್ಯಾರಾಮೀಟರ್‌ಗಳ ಪಟ್ಟಿಯನ್ನು ಹೊಂದಿರುವ ಅರೆ 


### ಫಂಕ್ಷನ್ ಕಾಲ್ ಮಾಡುವುದು 
ಫಂಕ್ಷನ್ ಅನ್ನು นิರುಪಿಸಿದ ನಂತರ, ನಾವು ಈಗ ಅದನ್ನು ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವ API ಕଲ್‌ಗೆ ಸೇರಿಸಬೇಕಾಗಿದೆ. ನಾವು ಇದನ್ನು ವಿನಂತಿಯಲ್ಲಿ `functions` ಅನ್ನು ಸೇರಿಸುವ ಮೂಲಕ ಮಾಡುವೆವು. ಈ ಪ್ರಕರಣದಲ್ಲಿ `functions=functions`.

`function_call` ಅನ್ನು `auto` ಗೆ ಸೆಟ್ ಮಾಡುವ ಆಯ್ಕೆಯೂ ಇದೆ. ಇದರ ಅರ್ಥ, ನಾವು ಸ್ವಯಂ ನಿರ್ಧರಿಸುವ ಬದಲು, ಬಳಕೆದಾರ ಸಂದೇಶದ ಆಧಾರದ ಮೇಲೆ ಯಾವ ಫಂಕ್ಷನ್ ಅನ್ನು ಕರೆದುಕೊಳ್ಳಬೇಕು ಎಂದು LLM ನಿರ್ಧರಿಸಲಿದೆ.

## 3.ಕಾರ್ಯಾಂಕ ವನ್ನು ಮನೋಧರ್ಮದಲ್ಲಿ ಕಾಂಪ್ಲಿಕೇಶನ್. 


ನಾವು LLM ನಿಂದ ಸ್ವರೂಪಗೊಳಿಸಿದ ಪ್ರತ್ಯುತ್ತರವನ್ನು ಪರೀಕ್ಷಿಸಿರುವ ನಂತರ, ನಾವು ಈಗ ಇದನ್ನು ಮನೋಧರ್ಮದಲ್ಲಿ ಸಂಯೋಜಿಸಬಹುದು. 

### ಪ್ರವಾಹವನ್ನು ನಿರ್ವಹಣೆ ಮಾಡುವುದು 

ಇದನ್ನು ನಮ್ಮ ಮನೋಧರ್ಮದಲ್ಲಿ ಸಂಯೋಜಿಸಲು, ಈ ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸೋಣ: 

ಮೊದಲು, OpenAI ಸೇವೆಗಳಿಗೆ ಕರೆ ಮಾಡಿ ಮತ್ತು ಸಂದೇಶವನ್ನು `response_message` ಎಂಬ ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸೋಣ. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ಈಗ ನಾವೀಗ ಮೈಸಕ್ರೋಸಾಫ್ಟ್ ಲರ್ನ್ API ಅನ್ನು ಕರೆಮಾಡಿ ಕೋರ್ಸುಗಳ ಪಟ್ಟಿ ಪಡೆಯಲು ಫಂಕ್ಶನ್ ಅನ್ನು ವ್ಯಾಖ್ಯಾನಿಸುವೆವು: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ಉತ್ತಮ ಅಭ್ಯಾಸವಾಗಿ, ನಾವು ಮೊದಲಿಗೆ ಮಾದರಿ ಯಾವುದೇ ಫಂಕ್ಷನ್ ಕರೆಸಿಕೊಳ್ಳಲು ಇಚ್ಛಿಸುತ್ತದೆಯೆಂದೂ ನೋಡೋಣ. ನಂತರ, ಲಭ್ಯವಿರುವ ಫಂಕ್ಷನ್‌ಗಳಲ್ಲಿ ಒಂದನ್ನು ರಚಿಸಿ ಇದಕ್ಕೆ ಕರೆಮಾಡಲಾಗುತ್ತಿರುವ ಫಂಕ್ಷನ್‌ಗೆ ಹೊಂದಿಸಲು ಕಂಡುಕೊಳ್ಳುತ್ತೇವೆ. 
ನಂತರ, ಫಂಕ್ಷನಿನ ಅರ್ಗ್ಯುಮೆಂಟ್ಸ್‌ಗಳನ್ನು ತೆಗೆದು LLM ನ ಅರ್ಗ್ಯುಮೆಂಟ್ಸ್‌ಗೆ ನಕ್ಷೇಪಣೆಯನ್ನು ಮಾಡುತ್ತೇವೆ.

ಕೊನೆಗೆ, ಫಂಕ್ಷನ್ ಕರೆ ಸಂದೇಶ ಮತ್ತು `search_courses` ಸಂದೇಶದಿಂದ ಹಿಂತಿರುಗಿಸಿದ ಮೌಲ್ಯಗಳನ್ನು ಸೇರಿಸುವೆವು. ಇದರಿಂದ LLMಗೆ ಬಳಕೆದಾರನಿಗೆ ಸಹಜ ಭಾಷೆಯಲ್ಲಿ ಪ್ರತಿಕ್ರಿಯಿಸುವ ಎಲ್ಲಾ ಮಾಹಿತಿಯೂ ಲಭ್ಯವಾಗುತ್ತದೆ.
ಬಳಕೆದಾರನಿಗೆ ಸಹಜ ಭಾಷೆ ಬಳಸಿಕೊಂಡು ಪ್ರತಿಕ್ರಿಯಿಸಲು LLMಗೆ ಅಗತ್ಯವಿರುವ ಎಲ್ಲಾ ಮಾಹಿತಿಯನ್ನು ನೀಡುತ್ತದೆ. 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ಈಗ ನಾವು ನವೀಕರಿಸಿದ ಸಂದೇಶವನ್ನು LLM ಗೆ ಕಳುಹಿಸುವೆವು ಆದ್ದರಿಂದ ನಾವು API JSON ಸ್ವರೂಪದ ಪ್ರತಿಕ್ರಿಯೆಯ ಬದಲಾಗಿ ಸಹಜ ಭಾಷೆ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪಡೆಯಬಹುದು.
