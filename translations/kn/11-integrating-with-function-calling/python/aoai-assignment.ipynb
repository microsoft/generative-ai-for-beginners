-Azure OpenAI ಬಳಸಿಕೊಂಡು ಕಾರ್ಯ ಕರೆ ಹೇಗಾಗಿಸುವುದು

## ಕಲಿಕೆ ಗುರಿಗಳು 

ಈ ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ಬಳಿಕ ನೀವು ತಿಳಿಯುತ್ತೀರಿ ಮತ್ತು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುತ್ತೀರಿ: 

- ಕಾರ್ಯ ಕರೆಗಳನ್ನು ಬಳಸದ ಮಗುವುದು ಯಾಕೆ 
- Azure Open AI ಸೇವೆಯನ್ನು ಬಳಸಿ ಕಾರ್ಯ ಕರೆ ಹೇಗೆ ಸೆಟಪ್ ಮಾಡುವುದು 
- ನಿಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ಬಳಕೆಕಾಗಿಗಾಗಿ ಪರಿಣಾಮಕಾರಿ ಕಾರ್ಯ ಕರೆಗಳನ್ನು ವಿನ್ಯಾಸ ಮಾಡುವುದು 


## ಫಂಕ್ಷನ್ ಕರೆಯುವಿಕೆಗಳ ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು 

ಈ ಪಾಠಕ್ಕಾಗಿ, ನಾವು ನಮ್ಮ ವಿದ್ಯಾ ಸ್ಟಾರ್ಟ್‌ ಅಪ್‌ಗೆ ಒಂದು ವೈಶಿಷ್ಟ್ಯವನ್ನು ನಿರ್ಮಿಸಲು ಬಯಸುತ್ತೇವೆ, ಅದು ಬಳಕೆದಾರರು ತಾಂತ್ರಿಕ ಕೋರ್ಸ್‌ಗಳನ್ನು ಕಂಡುಹಿಡಿಯಲು ಚಾಟ್‌ಬಾಟ್ ಅನ್ನು ಬಳಸಲು ಅನುವುಮಾಡುತ್ತದೆ. ನಾವು ಅವರ ಕೌಶಲ್ಯ ಮಟ್ಟ, ಪ್ರಸ್ತುತ ಹುದ್ದೆ ಮತ್ತು ಆಸಕ್ತಿ ಹೊಂದಿರುವ ತಂತ್ರಜ್ಞಾನದ ಆಧಾರವಾಗಿ ಕೋರ್ಸ್‌ಗಳನ್ನು ಶಿಫಾರಸು ಮಾಡುತ್ತೇವೆ. 

ಇದನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು ನಾವು ಬಳಸುವ ಸಂಯೋಜನೆ: 
 - `Azure Open AI` ಬಳಕೆದಾರರಿಗೆ ಚಾಟ್ ಅನುಭವವನ್ನು ರಚಿಸಲು
 - `Microsoft Learn Catalog API` ಬಳಕೆದಾರರ ವಿನಂತಿಯ ಆಧಾರದ ಮೇಲೆ ಕೋರ್ಸ್‌ಗಳನ್ನು ಹುಡುಕಲು ಸಹಾಯ ಮಾಡಲು
 - `Function Calling` ಬಳಕೆದಾರರ ಪ್ರಶ್ನೆಯನ್ನು ತೆಗೆದುಕೊಳ್ಳಲು ಮತ್ತು API ವಿನಂತಿಯನ್ನು ಮಾಡಲು ಫಂಕ್ಷನ್‌ಗೆ ಕಳುಹಿಸಲು. 

ಆರಂಭಿಸಲು, ನಾವು ಪ್ರಥಮವಾಗಿ ಫಂಕ್ಷನ್ ಕರೆಯುವಿಕೆಗಳನ್ನು ಯಾಕೆ ಬಳಸಬೇಕು ಎಂದು ನೋಡೋಣ: 


### ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಏಕೆ

ನೀವು ಈ ಕೋರ್ಸ್‌ನಲ್ಲಿ ಯಾವುದೇ ಇನ್ನೊಂದು ಪಾಠವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿರುವುದಾದರೆ, ನೀವು ಬಹುಶಃ ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗಳನ್ನು (LLMs) ಬಳಸುವ ಶಕ್ತಿಯನ್ನು ಅರ್ಥಮಾಡಿಕೊಂಡಿರಬಹುದು. ಹಾಗೆಯೇ, ನೀವು ಅವುಗಳ ಕೆಲವು ಮಿತಿಗಳನ್ನು ಸಹ ನೋಡಿ ಅರ್ಥಮಾಡಿಕೊಳ್ಳಬಹುದು ಎಂದು ನಿರೀಕ್ಷಿಸುತ್ತೇವೆ.

ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಎಂದರೆ, ಕೆಳಗಿನ ಮಿತಿಗಳನ್ನು ಮೀರುವ ಆಜೂರ್ ಓಪನ್ AI ಸೇವೆಯ ವೈಶಿಷ್ಟ್ಯ:
1) ಸತತ ಪ್ರತಿಕ್ರಿಯೆ ಸ್ವರೂಪ
2) ಚಾಟ್ ಸಂದರ್ಭಗಳಲ್ಲಿ ನಿರ್ಧಿಷ್ಟ ಅಪ್ಲಿಕೇಶನ್‌ನ ಇನ್ನೂ ಹೆಚ್ಚಿನ ಆಧಾರಗಳಿಂದ ಡೇಟಾವನ್ನು ಬಳಸುವ ಸಾಮರ್ಥ್ಯ

ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಮುಂಚೆ, LLMಗಳಿಂದ ಪ್ರತಿಕ್ರಿಯೆಗಳು ಅಸಂರಚಿತ ಹಾಗೂ ಅಸಮಾನವಾಗಿದ್ದವು. ಡೆವಲಪರ್‌ಗಳು ಪ್ರತಿಕ್ರಿಯೆಗಳ ಪ್ರತಿಯೊಂದನ್ನು ನಿರ್ವಹಿಸಲು ಸංකೀರ್ಣ ಪರಿಶೀಲನೆ ಕೋಡ್ ಬರೆದುಕೊಳ್ಳಬೇಕಾಗಿತ್ತು.

ಬಳಕೆದಾರರು "ಸ್ಟಾಕ್‌ಹೋಮ್‌ನಲ್ಲಿ ಪ್ರಸ್ತುತ ಹವಾಮಾನ ಏನು?" ಅಂತ ಮಾಹಿತಿ ಪಡೆಯಲಾಗುತ್ತಿರಲಿಲ್ಲ. ಇದಕ್ಕೆ ಕಾರಣ, ಮಾದರಿಗಳು ತರಬೇತುಗೊಂಡ ಡೇಟಾ ಸಮಯದ ಮಿತಿಯೊಳಗಿನವಷ್ಟೇ.

ಈ ಸಮಸ್ಯೆಯನ್ನು ವಿವರಿಸುವ ಕೆಳಗಿನ ಉದಾಹರಣೆಯೊಂದಕ್ಕೆ ನೋಡಿ:

ವಿದ್ಯಾರ್ಥಿಗಳ ಶ್ರೇಣಿಗಳನ್ನು ಸರಿಯಾದಂತೆ ಸೂಚಿಸಲು ನಾವು ವಿದ್ಯಾರ್ಥಿಗಳ ಡೇಟಾಬೇಸ್ ರಚಿಸಲಿಕ್ಕೆ ಬಯಸುತ್ತಿದ್ದೇವೆ ಎಂದು ನ предпಿ ಮಾಡೋಣ. ಕೆಳಗಿನ ಎರಡು ವಿದ್ಯಾರ್ಥಿಗಳ ವಿವರಗಳು ತುಂಬಾ ಸಮಾನವಾಗಿದೆ ಮತ್ತು ಅವುಗಳಲ್ಲಿನ ಡೇಟಾ ಸರಿಹೊಂದಿದೆ.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ನಾವು ಇದನ್ನು ಡೇಟಾವನ್ನು ವಿಶ್ಲೇಷಿಸಲು LLM ಗೆ ಕಳುಹಿಸಲು ಬಯಸುತ್ತೇವೆ. ಇದನ್ನು ನಂತರ ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್ ನಲ್ಲಿ API ಗೆ ಕಳುಹಿಸಲು ಅಥವಾ ಡೇಟಾಬೇಸಿನಲ್ಲಿ ಸಂಗ್ರಹಿಸಲು ಬಳಸಬಹುದು.

ನಾವು ಎರಡು ಸಮಾನವಾದ ಪ್ರಾಂಪ್ಟ್ಗಳನ್ನು ರಚಿಸೋಣ, ಈ ಮೂಲಕ ನಾವು LLM ಗೆ ನಾವು ಯಾವ ಮಾಹಿತಿಯಲ್ಲಿ ಆಸಕ್ತಿ ಹೊಂದಿದ್ದೇವೆ ಎಂದು ಹೇಳುತ್ತೇವೆ:


ಈ ಭಾಗಗಳನ್ನು ನಮ್ಮ ಉತ್ಪನ್ನಕ್ಕೆ ಮುಖ್ಯವಾಗಿರುವ ಅವುಗಳನ್ನು ವೀಕ್ಷಿಸಲು LLM ಗೆ ಕಳುಹಿಸಲು ನಾವು ಬಯಸುತ್ತೇವೆ. ಹೀಗಾಗಿ, ನಾವು LLM ಗೆ ಮಾರ್ಗದರ್ಶನ ಮಾಡಲು ಎರಡು ಒಂದೇ ತರಹದ ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ರಚಿಸಬಹುದು: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ಈ ಎರಡು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ರಚಿಸಿದ ನಂತರ, ನಾವು ಅವುಗಳನ್ನು `client.responses.create` ಬಳಸಿ LLM ಗೆ ಕಳುಹಿಸುತ್ತೇವೆ. ನಾವು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು `input` ಚರ변ಿಯಲ್ಲಿ ಸ೦ರಕ್ಷಿಸುತ್ತೇವೆ ಮತ್ತು ಪಾತ್ರವನ್ನು `user` ಗೆ ನಿಯೋಜಿಸುತ್ತೇವೆ. ಇದು ಬಳಕೆದಾರರಿಂದ ಸಂದೇಶವನ್ನು ಚಾಟ್‌ಬಾಟ್‌ಗೆ ಬರೆಯಲಾಗುತ್ತಿರುವಂತೆ ನಕಲಿ ಮಾಡಲು. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ಈಗ ನಾವು ಎರಡೂ ವಿನಂತಿಗಳನ್ನು LLM ಗೆ ಕಳುಹಿಸಿ ನಾವು ಪಡೆದ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪರಿಶೀಲಿಸಬಹುದು. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ಆದರೂ ಪ್ರಾಂಪ್ಟ್ಸ್ ಒಂದೇ ಇದ್ದರೂ ಮತ್ತು ವಿವರಗಳು ಹೋಲಿಕೆಯಾದರೂ, ನಾವು `Grades` ಗುಣಲಕ್ಷಣದ ವಿವಿಧ ಸ್ವರೂಪಗಳನ್ನು ಪಡೆಯಬಹುದು.

ಮೇಲಿನ ಸೆಲ್ ಅನ್ನು ಹಲವಾರು ಬಾರಿ ಚಾಲನೆ ಮಾಡಿದರೆ, ಸ್ವರೂಪವು `3.7` ಅಥವಾ `3.7 GPA` ಆಗಿರಬಹುದು.

ಇದು ಕಾರಣ LLM ಬರೆಯಲಾದ ಪ್ರಾಂಪ್ಟ್ ರೂಪದಲ್ಲಿರುವ ಅಸಂರಚಿತ ಡೇಟಾವನ್ನು ತೆಗೆದುಕೊಂಡು, ಮತ್ತೊಬ್ಬ ಅಸಂರಚಿತ ಡೇಟಾವನ್ನೇ ಮರಳುಕೊಳ್ಳುತ್ತದೆ. ನಾವು ಈ ಡೇಟಾವನ್ನು ಸಂಗ್ರಹಿಸುವಾಗ ಅಥವಾ ಬಳಕೆಮಾಡುವುದರಲ್ಲಿ ಯಾವ್ದನ್ನು ನಿರೀಕ್ಷಿಸುವುದು ಎಂಬ ತಿಳಿವಳಿಕೆಗಾಗಿ ಅದು ಸಂರಚಿತ ಸ್ವರೂಪದಲ್ಲಿ ಇರಬೇಕು.

ಕಾರ್ಯಾಪ್ರಕಾರಣ ಕರೆಯುವಿಕೆಯನ್ನು ಬಳಸಿ, ನಾವು ಸಂಕೇತಿತ ಡೇಟಾವನ್ನು ಪಡೆಯುವುದನ್ನು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಬಹುದು. ಕಾರ್ಯಾಪ್ರಕಾರಣ ಕರೆಯುವಿಕೆಯನ್ನು ಬಳಸಿದಾಗ, LLM ಸತ್ಯದಲ್ಲಿ ಯಾವುದೇ ಕಾರ್ಯಗಳನ್ನು ಕರೆಸುವುದಿಲ್ಲ ಅಥವಾ ಚಾಲನೆ ಮಾಡುವುದಿಲ್ಲ. ಬದಲಾಗಿ, ನಾವು LLM ಅದಕ್ಕೆ ಪಾಲಿಸಲು ಒಂದೇ ಸಂರಚನೆಯನ್ನು ಸೃಷ್ಟಿಸುತ್ತೇವೆ ಅದರ ಪ್ರತಿಕ್ರಿಯೆಗಾಗಿ. ಬಳಿಕ ಆ ಸಂಕೇತಿತ ಪ್ರತಿಕ್ರಿಯೆಗಳನ್ನು ಬಳಸಿ ನಮಗೆ ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ಗಳಲ್ಲಿ ಯಾವ ಕಾರ್ಯವನ್ನು ಚಾಲನೆ ಮಾಡಬೇಕು ಎಂದು ತಿಳಿಯುತ್ತದೆ.

ಬಳಕೆದಾರರು ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿಕೊಂಡು ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಿ, ಅದನ್ನು ಫಾರ್ಮ್ಯಾಟ್ ಮಾಡಿದ ಪ್ರಶ್ನೆ ಅಥವಾ API ವಿನಂತಿಯಾಗಿ ಪರಿವರ್ತಿಸಬಹುದು. ಉದಾಹರಣೆಗೆ, ಗುರು "ಕೊನೆಯ ಕಾರ್ಯವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ವಿದ್ಯಾರ್ಥಿಗಳು ಯಾರು?" ಎಂದು ಕೇಳಬಹುದು, ಇದು `get_completed(student_name: string, assignment: int, current_status: string)` ಎಂದು ಹೆಸರು ಮಾಡಲಾದ ಫಂಕ್ಷನ್ ಅನ್ನು ಕರೆ ಮಾಡಬಹುದು.  


**ರಚನಾತ್ಮಕ ಡೇಟಾವನ್ನು ರಚಿಸುವುದು**  
ಬಳಕೆದಾರರು ಪಠ್ಯದ ಒಂದು ಭಾಗ ಅಥವಾ CSV ಅನ್ನು ತೆಗೆದುಕೊಂಡು LLM ಬಳಸಿ ಅದರಲ್ಲಿರುವ ಮುಖ್ಯ ಮಾಹಿತಿಯನ್ನು ತೆಗೆಯಬಹುದು. ಉದಾಹರಣೆಗೆ, ವಿದ್ಯಾರ್ಥಿ ಶಾಂತಿ ಒಪ್ಪಂದಗಳ ಬಗ್ಗೆ ವಿಕಿಪೀಡಿಯಾ ಲೇಖನವನ್ನು AI ಫ್ಲಾಶ್ ಕಾರ್ಡ್ಸ್ ರಚಿಸಲು ಪರಿವರ್ತಿಸಬಹುದು. ಇದನ್ನು `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` ಎಂಬ ಫಂಕ್ಷನ್ ಕರೆಯುವ ಮೂಲಕ ಮಾಡಬಹುದು.  


## 2. ನಿಮ್ಮ ಪ್ರಥಮ ಫಂಕ್ಷನ್ ಕರೆಗೆ ಸೃಷ್ಟಿಸುವುದು 

ಫಂಕ್ಷನ್ ಕರೆಗೆ ಸೃಷ್ಟಿಸುವ ಪ್ರಕ್ರಿಯೆ 3 ಪ್ರಮುಖ ಹಂತಗಳನ್ನು ಒಳಗೊಂಡಿದೆ: 
1. ನಿಮ್ಮ ಫಂಕ್ಷನ್‌ಗಳ ಪಟ್ಟಿ ಮತ್ತು ಬಳಕೆದಾರ ಸಂದೇಶದೊಂದಿಗೆ ಚಾಟ್ ಸಮಾಪ್ತಿಗಳ API ಅನ್ನು ಕರೆಸುವುದು 
2. ಕ್ರಿಯೆ ಅಥವಾ API ಕರೆಗಾಗಿ ಫಂಕ್ಷನ್ ಅನ್ನು ನಿರ್ವಹಿಸಲು ಮಾದರಿಯ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಓದುವುದು 
3. ಬಳಕೆದಾರನಿಗೆ ಪ್ರತಿಕ್ರಿಯೆ ಸೃಷ್ಟಿಸಲು ನಿಮ್ಮ ಫಂಕ್ಷನ್‌ನ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಬಳಸಿ ಚಾಟ್ ಸಮಾಪ್ತಿಗಳ API ಅನ್ನು ಮತ್ತೊಮ್ಮೆ ಕರೆಸುವುದು.

### ಕಾರ್ಯಾಚರಣೆ ಕರೆ ಸಹಾಯಕಾಂಶಗಳು 

#### ಬಳಕೆದಾರರ ಇನ್ಪುಟ್ 

ಮೊದಲನೆಯ ಹಂತವೆಂದರೆ ಬಳಕೆದಾರ ಸಂದೇಶವನ್ನು ರಚಿಸುವುದು. ಇದನ್ನು ಪಠ್ಯ ಇನ್ಪುಟ್‌ನ ಮೌಲ್ಯ ಪಡೆಯುವುದರಿಂದ ಗತಿಶೀಲವಾಗಿ ನಿಯೋಜಿಸಬಹುದು ಅಥವಾ ನೀವು ಇಲ್ಲಿ ಮೌಲ್ಯವನ್ನು ನಿಯೋಜಿಸಬಹುದು. ನೀವು ಚಾಟ್ কমಪ್ಲೀಶನ್ಸ್ API ಜೊತೆಗೆ ಮೊದಲ ಬಾರಿಗೆ ಕೆಲಸ ಮಾಡುತ್ತಿದ್ದರೆ, ಸಂದೇಶದ `role` ಮತ್ತು `content` ಅನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಬೇಕಾಗುತ್ತದೆ. 

`role` ಯಾದರೂ `system` (ನಿಯಮಗಳನ್ನು ರಚಿಸುವುದು), `assistant` (ಮಾದರಿ) ಅಥವಾ `user` (ಅಂತಿಮ ಬಳಕೆದಾರ) ಆಗಿರಬಹುದು. ಕಾರ್ಯಾಚರಣೆ ಕರೆಗಾಗಿ, ನಾವು ಇದನ್ನು `user` ಮತ್ತು ಒಂದು ಉದಾಹರಣಾ ಪ್ರಶ್ನೆಯಾಗಿ ನಿಯೋಜಿಸುತ್ತೇವೆ. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ಕಾರ್ಯಗಳನ್ನು ರಚಿಸುವುದು. 

ಮುಂದಿಗೆ ನಾವು ಒಂದು ಕಾರ್ಯವನ್ನು ಮತ್ತು ಆ ಕಾರ್ಯದ ಪರಿಮಾಣಗಳನ್ನು ವ್ಯಾಖ್ಯಾನಿಸುವೆವು. ನಾವು ಇಲ್ಲಿ `search_courses` ಎಂಬ ಒಂದು ಕಾರ್ಯವನ್ನು ಮಾತ್ರ ಬಳಸುತ್ತೇವೆ ಆದರೆ ನೀವು ಬಹು ಕಾರ್ಯಗಳನ್ನು ರಚಿಸಬಹುದು.

**ಮುಖ್ಯ** : ಕಾರ್ಯಗಳು LLM ಗೆ ನೀಡುವ ಸಿಸ್ಟಮ್ ಸಂದೇಶದಲ್ಲಿ ಸೇರುತ್ತವೆ ಮತ್ತು ಅವು ನಿಮ್ಮ ಲಭ್ಯವಿರುವ ಟೋಕನ್‌ಗಳ ಪ್ರಮಾಣದಲ್ಲಿ ಸೇರಿಸಲ್ಪಡುತ್ತವೆ. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**ವ್ಯಾಖ್ಯಾನಗಳು** 

`name` - ನಾವು ಕರೆ ಮಾಡಬೇಕಾದ ಫಂಕ್ಷನ್‌ನ ಹೆಸರು. 

`description` - ಫಂಕ್ಷನ್ ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ ಎಂಬ ವಿವರ. ಇಲ್ಲಿ ಸ್ಪಷ್ಟ ಮತ್ತು ನಿಖರವಾಗಿರುವುದು ಮುಖ್ಯ. 

`parameters` - ನೀವು ಮಾದರಿಯಿಂದ ತನ್ನ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಉತ್ಪತ್ತಿಯಾಗಬೇಕಾದ ಮೌಲ್ಯಗಳು ಮತ್ತು ಆಕಾರದ ಪಟ್ಟಿಯನ್ನು ನೀಡುವುದು. 


`type` - ನಿರ್ದಿಷ್ಟ ಲಕ್ಷಣಗಳ ಡೇಟಾ ಪ್ರಕಾರವನ್ನು ಇಲ್ಲಿ ಸಂಗ್ರಹಿಸಲಾಗುತ್ತದೆ. 

`properties` - ಮಾದರಿ ತನ್ನ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಬಳಸದಿರುವ ನಿಗದಿತ ಮೌಲ್ಯಗಳ ಪಟ್ಟಿಯನ್ನು ಸೂಚಿಸುತ್ತದೆ. 


`name` - ಮಾದರಿ ತನ್ನ ಸೂಕ್ತ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಬಳಸುವ ಲಕ್ಷಣದ ಹೆಸರು. 

`type` - ಈ ಲಕ್ಷಣದ ಡೇಟಾ ಪ್ರಕಾರ. 

`description` - ನಿರ್ದಿಷ್ಟ ಲಕ್ಷಣದ ವಿವರಣೆ. 


**ಐಚ್ಛಿಕ** 

`required` - ಫಂಕ್ಷನ್ ಕರೆ ಪೂರ್ಣಗೊಳ್ಳಲು ಅಗತ್ಯವಿರುವ ಲಕ್ಷಣ. 


### ಕಾರ್ಯವನ್ನು ಕರೆಯುವುದು  
ಒಂದು ಕಾರ್ಯವನ್ನು ವ್ಯಾಖ್ಯಾನಿಸಿದ ನಂತರ, ನಾವು ಅದನ್ನು Chat Completion APIಗೆ ಕರೆ ಮಾಡಲು ಸೇರಿಸಬೇಕಾಗಿದೆ. ಇದಕ್ಕಾಗಿ ನಾವು ವಿನಂತಿಗೆ `functions` ಅನ್ನು ಸೇರಿಸುತ್ತೇವೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ `functions=functions` ಆಗಿದೆ.  

`function_call` ಅನ್ನು `auto` ಗೆ ಹೊಂದಿಸುವ ಆಯ್ಕೆಯನ್ನು ಕೂಡಾ ಇದೆ. ಇದರ ಅರ್ಥ, ನಾವು ಸ್ವಯಂ ಕಾರ್ಯವನ್ನು ನಿಯೋಜಿಸುವ ಬದಲಾಗಿ ಬಳಕೆದಾರ ಸಂದೇಶ ಆಧಾರದ ಮೇಲೆ LLM ಯಾವ ಕಾರ್ಯವನ್ನು ಕರೆಮಾಡಬೇಕು ಎಂದು ತೀರ್ಮಾನಿಸುವ ಅವಕಾಶ ನೀಡುತ್ತೇವೆ.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ಈಗ ಮೊದಲು ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ನೋಡೋಣ ಮತ್ತು ಅದು ಹೇಗೆ ವಿನ್ಯಾಸಗೊಳ್ಳಲಾಗಿದೆ ಎನ್ನುತ್ತದೆ: 

{
  "role": "assistant",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

ನೀವು ಗಮನಿಸಬಹುದು, function ನ ಹೆಸರು ಕರೆ ಮಾಡಲಾಗಿದೆ ಮತ್ತು ಬಳಕೆದಾರರ ಸಂದೇಶದಿಂದ, LLM function ನ arguments ಗೆ ಹೊಂದಿಕೊಳ್ಳುವ ಡೇಟಾವನ್ನು ಕಂಡುಹಿಡಿಯಲು ಸಾಧ್ಯವಾಗಿದೆ. 


## 3.ಅನ್ವಯದಲ್ಲಿ ಫಂಕ್ಷನ್ ಕಾಲ್ಗಳನ್ನು ಏಕೀಕರಿಸುವುದು.


ನಾವು LLM ನಿಂದ ಸ್ವರೂಪಿತ ಪ್ರತ್ಯುತ್ತರವನ್ನ ಪರೀಕ್ಷಿಸಿದ ಮೇಲೆ, ಈಗ ಅದನ್ನು ಅನ್ವಯದಲ್ಲಿ ಏಕೀಕರಿಸಬಹುದು.

### ಹರಿವನ್ನು ನಿರ್ವಹಿಸುವುದು

ಇದನ್ನು ನಮ್ಮ ಅನ್ವಯದಲ್ಲಿ ಏಕೀಕರಿಸಲು, ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ತೆಗೆದುಕೊಳ್ಳೋಣ:

ಮೊದಲು, Open AI ಸೇವೆಗಳಿಗೆ ಕರೆ ಮಾಡಿ ಮತ್ತು ಸಂದೇಶವನ್ನು `response_message` ಎಂಬ ಚರದಲ್ಲಿ ಸಂಗ್ರಹಿಸೋಣ.


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ಈಗ ನಾವು Microsoft Learn API ಅನ್ನು ಕರೆಮಾಡುವ, ಕೋರ್ಸ್‌ಗಳ ಪಟ್ಟಿಯನ್ನು ಪಡೆಯುವ ಫಂಕ್ಷನ್ ಅನ್ನು ನಿರ್ಧರಿಸುವೆವು: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ಉತ್ತಮ ಪ್ರ್ಯಾಕ್ಟೀಸ್ ಆಗಿ, ನಂತರ ನಾವು ಮಾದರಿ ಯಾವುದೇ ಫಂಕ್ಷನ್ ಕರೆ ಮಾಡಲು ಇಚ್ಛಿಸುವದನ್ನು ನೋಡುತ್ತೇವೆ. ಅದರಿಂದ ನಂತರ, ಲಭ್ಯವಿರುವ ಫಂಕ್ಷನ್‌ಗಳಲ್ಲಿ ಒಂದನ್ನು ರಚಿಸಿ ಮತ್ತು ಕರೆ ಮಾಡಲಾಗುತ್ತಿರುವ ಫಂಕ್ಷನ್‌ಗೆ ಹೊಂದಿಸುವೆವು.
ನಂತರ ಫಂಕ್ಷನ್‌ನ арг್ಯುಮೆಂಟ್‌ಗಳನ್ನು ತೆಗೆದು LLM‌ನ арг್ಯುಮೆಂಟ್‌ಗಳಿಗೆ ನಕ್ಷೆ ಮಾಡುತ್ತೇವೆ.

ಕೊನೆಯದಾಗಿ, ನಾವು ಫಂಕ್ಷನ್ ಕಲ್ ಸಂದೇಶ ಮತ್ತು `search_courses` ಸಂದೇಶದಿಂದ ಪ್ರತಿಕ್ರಿಯಿಸಿದ ಮೌಲ್ಯಗಳನ್ನು ಸೇರಿಸುತ್ತೇವೆ. ಇದರಿಂದ LLMಗೆ
ಸ್ವಾಭಾವಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಬಳಕೆದಾರರಿಗೆ ಪ್ರತಿಕ್ರಿಯಿಸುವ ಎಲ್ಲಾ ಮಾಹಿತಿಯನ್ನು ದೊರಕುತ್ತದೆ.


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ಇದೀಗ ನಾವು ನವೀಕರಿಸಲಾದ ಸಂದೇಶವನ್ನು LLM ಗೆ ಕಳುಹಿಸುವೆವು ಆದ್ದರಿಂದ ನಾವು API JSON ಸ್ವರೂಪಿತ ಉತ್ತರದ ಬದಲಾವಣೆಗೆ ಸ್ವಾಭಾವಿಕ ಭಾಷೆಯ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪಡೆಯಬಹುದು.
