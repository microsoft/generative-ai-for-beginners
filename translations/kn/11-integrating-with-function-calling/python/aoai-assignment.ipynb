## ಕಾರ್ಯ ಕಾಲ್‌ಗಳನ್ನು ಅರ್ಥಮಾಡಿಕೊಳ್ಳುವುದು

ಈ ಪಾಠಕ್ಕಾಗಿ, ನಾವು ನಮ್ಮ ಶಿಕ್ಷಣ ಸ್ಟಾರ್ಟ್‌ಅಪ್‌ಗೆ ಒಂದು ವೈಶಿಷ್ಟ್ಯವನ್ನು ನಿರ್ಮಿಸಬಯಸುತ್ತೇವೆ, ಇದು ಬಳಕೆದಾರರಿಗೆ ಟೆಕ್ನಿಕಲ್ ಕೋರ್ಸುಗಳನ್ನು ಹುಡುಕಲು ಚಾಟ್‌ಬಾಟ್ ಬಳಸಲು ಅನುಮತಿಸುತ್ತದೆ. ನಾವು ಅವರ ಕೌಶಲ್ಯ ಮಟ್ಟ, ಪ್ರಸ್ತುತ ಪಾತ್ರ ಮತ್ತು ಆಸಕ್ತಿಯ ತಂತ್ರಜ್ಞಾನಕ್ಕೆ ಹೊಂದಿಕೆಯಾಗುವ ಕೋರ್ಸುಗಳನ್ನು ಶಿಫಾರಸು ಮಾಡುತ್ತೇವೆ.

ಇದನ್ನು ಪೂರ್ಣಗೊಳಿಸಲು, ನಾವು ಕೆಳಗಿನ ಸಂಯೋಜನೆಯನ್ನು ಬಳಸುತ್ತೇವೆ:
 - ಬಳಕೆದಾರರಿಗೆ ಚಾಟ್ ಅನುಭವವನ್ನು ರಚಿಸಲು `Azure Open AI`
 - ಬಳಕೆದಾರರಿಂದ ಕೇಳಿದ ವಿನಂತಿಯ ಆಧಾರದ ಮೇಲೆ ಕೋರ್ಸುಗಳನ್ನು ಹುಡುಕಲು ಸಹಾಯ ಮಾಡುವ `Microsoft Learn Catalog API`
 - ಬಳಕೆದಾರರ ಪ್ರಶ್ನೆಯನ್ನು ಪಡೆದು API ವಿನಂತಿಯನ್ನು ಮಾಡಲು ಅದನ್ನು ಕಾರ್ಯಕ್ಕೆ ಕಳುಹಿಸುವ `Function Calling`

ಪ್ರಾರಂಭಿಸಲು, ನಾವು ಮೊದಲು ಕಾರ್ಯ ಕಾಲ್ ಅನ್ನು ನಮ್ಮೆಂತಹ ಬಳಸಬೇಕೆಂದು ನೋಡೋಣ:


### ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆ ಯಾಕೆ 

ನೀವು ಈ ಪಠ್ಯಕ್ರಮದಲ್ಲಿ ಇನ್ನೊಂದಿಷ್ಟು ಪಾಠ ಮುಗಿಸಿರಬಹುದೇ, ನೀವು ದೊಡ್ಡ ಭಾಷಾ ಮಾದರಿಗಳ (LLMs) ಬಳಕೆಯ ಶಕ್ತಿಯನ್ನು ಅರ್ಥಮಾಡಿಕೊಂಡಿರಬಹುದು. ಆಶಾಸ್ಪದವಾಗಿ ನೀವು ಅದರ ಕೆಲವು ಮಿತಿಗಳನ್ನು ಕೂಡ ನೋಡಬಹುದು. 

ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆ ಎಂಬುದು ಅಜೂರ್ ಓಪನ್ ಏಐ ಸೇವೆಯ ವೈಶಿಷ್ಟ್ಯ, ಇದು ಕೆಳಗಿನ ಮಿತಿಗಳನ್ನು ಮೀರುವುದು: 
೧) ಸವಾಲುಪಾತ್ರವಾದ ಪ್ರತಿಕ್ರಿಯೆಯ ಶೈಲಿ 
೨) ಚಾಟ್ ಪರಿಸರದಲ್ಲಿ ಅನ್ವಯದ ಇತರೆ ಮೂಲಗಳ ಡೇಟಾವನ್ನು ಬಳಕೆಮಾಡುವ ಸಾಮರ್ಥ್ಯ 

ಫಂಕ್ಷನ್ ಕರೆಮಾಡುವಿಕೆಯ ಮೊದಲು, LLM ನಿಂದ ಬಂದ ಪ್ರತಿಕ್ರಿಯೆಗಳು ಅಸಂರಚಿತ ಮತ್ತು ಅಸಮರ್ಪಕವಾಗಿದ್ದವು. ಡೆವಲಪರ್‌ಗಳು ಪ್ರತಿಯೊಂದು ಪ್ರತಿಕ್ರಿಯೆಯ ಬದಲಾಗುವುದನ್ನು ನಿರ್ವಹಿಸಲು ಸಂಕೀರ್ಣ ಪರಿಶೀಲನೆ ಕೋಡ್ ಬರೆಯಬೇಕಾಗಿತ್ತು. 

"ಸ್ಟಾಕ್ಹೋಲ್ಮ್ ನಲ್ಲಿ ಪ್ರಸ್ತುತ ಹವಾಮಾನ ಏನು?" ಎಂಬಂತಹ ಉತ್ತರಗಳನ್ನು ಬಳಕೆದಾರರು ಪಡೆಯಲಾರರು. ಇದಕ್ಕೆ ಕಾರಣ ಮಾದರಿಗಳು ತರಬೇತಿಗೊಂಡಿದ್ದ ಡೇಟಾದ ಸಮಯಕ್ಕೆ ಸೀಮಿತವಾಗಿದ್ದವು. 

ಈ ಸಮಸ್ಯೆಯನ್ನು ವಿವರಿಸುವ ಕೆಳಗಿನ ಉದಾಹರಣೆಯನ್ನು ನೋಡೋಣ: 

ನಾವು ವಿದ್ಯಾರ್ಥಿಗಳ ಡೇಟಾಬೇಸ್ ಸೃಷ್ಟಿಸಲು ಬಯಸಿದರೆ, ಅವರಿಗೆ ಸರಿಯಾದ ಕೋರ್ಸನ್ನು ಸಲಹೆ ನೀಡಲು ಸಹಾಯವಾಗುತ್ತದೆ. ಕೆಳಗೆ ನಮಗೆ ಎರಡು ವಿದ್ಯಾರ್ಥಿಗಳ ವಿವರಣೆಗಳಿವೆ, ಅವು ಡೇಟಾದಲ್ಲಿಯೂ ಬಹಳ ಸಮಾನವಾಗಿವೆ.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ನಾವು ಈ ಮಾಹಿತಿಯನ್ನು LLM ಗೆ ಕಳುಹಿಸಿ, ದತ್ತಾಂಶವನ್ನು ವಿಶ್ಲೇಷಿಸಲು ಇಚ್ಛಿಸುತ್ತೇವೆ. ಇದನ್ನು ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ನಂತರ API ಗೆ ಕಳುಹಿಸಬಹುದು ಅಥವಾ ಡೇಟಾಬೇಸ್‌ನಲ್ಲಿ ಸಂಗ್ರಹಿಸಬಹುದು. 

ನಾವು ಎರಡು ಸಮಾನವಾದ ಪ್ರಾಂಪ್ಟುಗಳನ್ನು ಸೃಷ್ಠಿಸೋಣ, ಅವುಗಳಲ್ಲಿ ನಾವು LLM ಗೆ ನಾವು ಯಾವ ಮಾಹಿತಿಯದಲ್ಲಿ ಆಸಕ್ತಿ ಹೊಂದಿದ್ದೇವೆ ಎಂದು ನಿರ್ದಿಷ್ಟಪಡಿಸುತ್ತೇವೆ: 


ನಾವು ನಮ್ಮ ಉತ್ಪನ್ನಕ್ಕೆ ಮುಖ್ಯವಾದ ಭಾಗಗಳನ್ನು ವಿಶ್ಲೇಷಿಸಲು ಇದನ್ನು LLM ಗೆ ಕಳುಹಿಸಲು ಇಚ್ಛಿಸುತ್ತೇವೆ. ಆದ್ದರಿಂದ ನಾವು LLM ಗೆ ಸೂಚಿಸಲು ಎರಡು ಒಂದೇ_PROMPTS ಅನ್ನು ಸೃಷ್ಟಿಸಬಹುದು: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ಈ ಎರಡು ಪ್ರಾಂಪ್ಟ್‌ಗಳನ್ನು ರಚಿಸಿದ ನಂತರ, ನಾವು ಅವನ್ನು LLMಗೆ `client.responses.create` ಬಳಸಿ ಕಳುಹಿಸುವೆವು. ನಾವು ಪ್ರಾಂಪ್ಟ್ ಅನ್ನು `input` ವೇರಿಯಬಲ್‌ನಲ್ಲಿ ಸಂಗ್ರಹಿಸಿ, ಪಾತ್ರವನ್ನು `user` ಗೆ ಹಂಚುವೆವು. ಇದು ಬಳಕೆದಾರನಿಂದ ಬಂದ ಸಂದೇಶವನ್ನು ಚಾಟ್‌ಬಾಟ್‌ಗೆ ಬರೆಯುತ್ತಿರುವಂತೆ ಅನುಕರಿಸಲು. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ಈಗ ನಾವು ಎರಡು ವಿನಂತಿಗಳನ್ನು LLM ಗೆ ಕಳುಹಿಸಿ ನಾವು ಹೊಂದುವ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪರಿಶೀಲಿಸಬಹುದು. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ಪ್ರಾಂಪ್ಟ್‌ಗಳು ಒಂದೇ ಆಗಿದ್ದರೂ ಹಾಗೂ ವಿವರಣೆಗಳು ಸಮಾನವಾಗಿದ್ದರೂ, ನಾವು `Grades` ಗುಣಲಕ್ಷಣದ ವಿಭಿನ್ನ ಫಾರ್ಮ್ಯಾಟ್ಗಳನ್ನು ಪಡೆಯಬಹುದು.

ಇದು ಕಾರಣವೆಂದರೆ LLM ಬರೆಯಲಾದ ಪ್ರಾಂಪ್ಟ್ ರೂಪದಲ್ಲಿ ಅಸಂರಚಿತ ಡೇಟಾವನ್ನು ತೆಗೆದುಕೊಳ್ಳುತ್ತದೆ ಮತ್ತು ಅಸಂರಚಿತ ಡೇಟಾನ್ನೇ ಹಿಂತಿರುಗಿಸುತ್ತದೆ. ನಾವು ಸಂರಚಿತ ಫಾರ್ಮ್ಯಾಟ್ ಅನ್ನು ಹೊಂದಿರಬೇಕು, ಹಾಗೆಯೇ ನಾವು ಈ ಡೇಟಾವನ್ನು ಸಂಗ್ರಹಿಸುವಾಗ ಅಥವಾ ಬಳಸುವಾಗ ಏನು ಎದುರು ನೋಡಬಹುದು ಎನ್ನುವುದನ್ನು ಗೊತ್ತುಮಾಡಿಕೊಳ್ಳಲು. 

ಫಂಕ್ಷನಲ್ ಕಾಲಿಂಗ್ ಬಳಸಿ, ನಾವು ಸಂರಚಿತ ಡೇಟಾವನ್ನು ಹಿಂತಿರುಗಿಸುವುದಾಗಿ ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಬಹುದು. ಫಂಕ್ಷನ್ ಕಾಲಿಂಗ್ ಬಳಸದಾಗಲೂ LLM ನಿಷ್ಠೆ, ಯಾವುದೇ ಫಂಕ್ಷನ್‌ಗಳನ್ನು ನೇರವಾಗಿ ಕರೆ ಮಾಡದು ಅಥವಾ ನಿರ್ವಹಿಸುವುದಿಲ್ಲ. ಬದಲಾಗಿ, ಪ್ರತಿಕ್ರಿಯೆಗಳಿಗೆ ಅನುಸರಿಸಲು LLMಗೆ ಒಂದು ಸಂರಚಿತ ರಚನೆಯನ್ನು ನಾವು ಸೃಷ್ಟಿಸುತ್ತೇವೆ. ನಂತರ ನಾವು ಆ ಸಂರಚಿತ ಪ್ರತಿಕ್ರಿಯೆಗಳನ್ನು ತನ್ನ ಪ್ರಕಾರ ನಮಗೆ ಯಾವ ಫಂಕ್ಷನ್ ಬಳಸಿ ನಿರ್ವಹಣೆ ಮಾಡಬೇಕು ಎಂದು ತಿಳಿದುಕೊಳ್ಳಲು ಬಳಸುತ್ತೇವೆ.

ಬಳಕೆದಾರರು ನೈಸರ್ಗಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಮಾಹಿತಿಯನ್ನು ಹುಡುಕಬಹುದು, ಇದು ರೂಪುಗೊಂಡ ವಿನಂತಿ ಅಥವಾ API ವಿನಂತಿಯಾಗಿ ಪರಿವರ್ತಿತವಾಗುತ್ತದೆ. ಉದಾಹರಣೆಗೆ, ಒಂದು ಶಿಕ್ಷಕ "ಕೊನೆಯ ಕಾರ್ಯವನ್ನು ಪೂರ್ಣಗೊಳಿಸಿದ ವಿದ್ಯಾರ್ಥಿಗಳು ಯಾರು" ಎಂದು ಕೇಳಿದರೆ, ಅದು `get_completed(student_name: string, assignment: int, current_status: string)` ಎಂಬ ಕಾರ್ಯವನ್ನು ಕರೆ ಮಾಡಬಹುದು  


**ಸಂರಚಿತ ಡೇಟಾವನ್ನು ರಚಿಸುವುದು**
ಬಳಕೆದಾರರು ಪಠ್ಯ ಅಥವಾ CSV ಬ್ಲಾಕ್‌ನಿಂದ ಐತಿಹಾಸಿಕ ಮಾಹಿತಿಯನ್ನು LLM ಮೂಲಕ ಹೊರತೆಗೆಯಬಹುದು. ಉದಾಹರಣೆಗೆ, ವಿದ್ಯಾರ್ಥಿ ಶಾಂತಿ ಒಪ್ಪಂದಗಳ ಬಗ್ಗೆ ವಿಕಿಪೀಡಿಯ ಲೇಖನವನ್ನು AI ಫ್ಲಾಷ್ ಕಾರ್ಡುಗಳನ್ನು ರಚಿಸಲು ಪರಿವರ್ತಿಸಬಹುದು. ಇದನ್ನು `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` ಎಂಬ ಕಾರ್ಯವನ್ನು ಬಳಸಿ ಮಾಡಬಹುದು  


## 2. ನಿಮ್ಮ ಮೊದಲ ಫಂಕ್ಷನ್ কল್ ರಚಿಸುವುದು 

ಫಂಕ್ಶನ್ কল್ ರಚಿಸುವ ಪ್ರಕ್ರಿಯೆ 3 ಪ್ರಮುಖ ಹಂತಗಳನ್ನು ಒಳಗೊಂಡಿದೆ: 
1. ನಿಮ್ಮ ಫಂಕ್ಷನ್ಗಳ ಪಟ್ಟಿಯೊಂದಿಗೆ ಮತ್ತು ಬಳಕೆದಾರ ಸಂದೇಶದೊಂದಿಗೆ ಚಾಟ್ ಕಂಪ್ಲೀಶನ್ಸ್ API ಅನ್ನು ಕರೆಮಾಡುವುದು 
2. ಕಾರ್ಯಾಚರಣೆ ಮಾಡಲು ಮಾದರಿಯ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಓದುವುದು ಅಂದರೆ ಫಂಕ್ಷನ್ ಅಥವಾ API ಕಾಲ್ ಅನ್ನು ನಿರ್ವಹಿಸುವುದು 
3. ಬಳಕೆದಾರರಿಗೆ ಪ್ರತಿಕ್ರಿಯೆ ರಚಿಸಲು ಆ ಮಾಹಿತಿಯನ್ನು ಬಳಸುವ ನಿಮ್ಮ ಫಂಕ್ಷನ್‌ನ ಪ್ರತಿಕ್ರಿಯೆಯೊಂದಿಗೆ ಮತ್ತೊಂದು ಚಾಟ್ ಕಂಪ್ಲೀಶನ್ಸ್ API ಅನ್ನು ಕರೆಮಾಡುವುದು.

#### ಬಳಕೆದಾರರ ಇನ್ಪುಟ್ 

ಮೊದಲನೆಯ ಹಂತವು ಬಳಕೆದಾರ ಸಂದೇಶವನ್ನು ರಚಿಸುವುದು. ಈ ಸಂದೇಶವನ್ನು ಪಠ್ಯ ಇನ್ಪುಟ್‌ನ ಮೌಲ್ಯವನ್ನು ಪಡೆದು ಡೈನಮಿಕಾಗಿ ನಿಯೋಜಿಸಬಹುದಾಗಿದೆ ಅಥವಾ ನೀವು ಇಲ್ಲಿ ಮೌಲ್ಯವನ್ನು ನಿಯೋಜಿಸಬಹುದು. ನೀವು ಚಾಟ್ ಪೂರ್ಣಗೊಳಿಸುವಿಕೆ API-ಯೊಂದಿಗೆ ಮೊದಲ ಬಾರಿಗೆ ಕೆಲಸ ಮಾಡುತ್ತಿದ್ದರೆ, ನಾವು ಸಂದೇಶದ `role` ಮತ್ತು `content` ಅನ್ನು ನಿರ್ದಿಷ್ಟಗೊಳಿಸಬೇಕಾಗಿದೆ. 

`role` ಆಗಿರಬಹುದಾದವು `system` (ನಿಯಮಗಳನ್ನು ರಚಿಸುವುದು), `assistant` (ಮಾದರಿ), ಅಥವಾ `user` (ಅಂತಿಮ ಬಳಕೆದಾರ). ಕಾರ್ಯದ ಕರೆಯಿಗಾಗಿ, ನಾವು ಇದನ್ನು `user` ಎಂದು ನಿಯೋಜಿಸುವೆವು ಮತ್ತು ಉದಾಹರಣೆಯ ಪ್ರಶ್ನೆಯನ್ನು ನೀಡುವೆವು. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ಫಂಕ್ಷನ್‌ಗಳನ್ನು ರಚಿಸುವುದು. 

ಮುಂದಿನದಾಗಿ ನಾವು ಒಂದು ಫಂಕ್ಷನ್ ಮತ್ತು ಆ ಫಂಕ್ಷನಿನ ಪ್ಯಾರಾಮೀಟರ್‌ಗಳನ್ನು ವ್ಯಾಖ್ಯಾನಿಸೋಣ. ಇಲ್ಲಿ ನಾವು ಒಂದೇ ಫಂಕ್ಷನ್ ಅನ್ನು ಬಳಸುವೆವು, ಅದನ್ನು `search_courses` ಎನ್ನುತ್ತೇವೆ ಆದರೆ ನೀವು ಬಹು ಫಂಕ್ಷನ್‌ಗಳನ್ನು ರಚಿಸಬಹುದು.

**ಮಹತ್ವದ:** ಫಂಕ್ಷನ್‌ಗಳು LLM ಗೆ ಸಿಸ್ಟಮ್ ಸಂದೇಶದಲ್ಲಿ ಸೇರಿಸಲಾಗುತ್ತದೆ ಮತ್ತು ನಿಮ್ಮ ಬಳಕೆಯಲ್ಲಿರುವ ಟೋಕನ್‌ಗಳ ಪ್ರಮಾಣದಲ್ಲಿ ಅದರ ವಿಸ್ತಾರವೂ ಸೇರಿರುತ್ತದೆ. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**ವ್ಯಾಖ್ಯಾನಗಳು** 

`name` - ನಾವು ಕರೆಸಬೇಕಾದ ಫಂಕ್ಷನ್‌ಗಾಗಿ ಹೆಸರು. 

`description` - ಫಂಕ್ಷನ್ ಹೇಗೆ ಕಾರ್ಯನಿರ್ವಹಿಸುತ್ತದೆ ಎಂಬ ವಿವರಣೆ. ಇಲ್ಲಿ ಸ್ಪಷ್ಟ ಮತ್ತು ನಿಖರವಾಗಿರುವುದು ಮುಖ್ಯ. 

`parameters` - ಮಾದರಿ ತನ್ನ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಉತ್ಪಾದಿಸಬೇಕಾದ ಮೌಲ್ಯಗಳು ಮತ್ತು ಫಾರ್ಮ್ಯಾಟ್‌ಗಳ ಪಟ್ಟಿ 


`type` - ಗುಣಲಕ್ಷಣಗಳನ್ನು ಸಂಗ್ರಹಿಸಬೇಕಾದ ಡೇಟಾ ಪ್ರಕಾರ 

`properties` - ಮಾದರಿ ತನ್ನ ಪ್ರತಿಕ್ರಿಯೆಗಾಗಿ ಬಳಸಲಿರುವ ನಿರ್ದಿಷ್ಟ ಮೌಲ್ಯಗಳ ಪಟ್ಟಿ 


`name` - ಮಾದರಿ ತನ್ನ ಫಾರ್ಮ್ಯಾಟ್ ಮಾಡಿದ ಪ್ರತಿಕ್ರಿಯೆಯಲ್ಲಿ ಬಳಸಲಿರುವ ಗುಣಲಕ್ಷಣದ ಹೆಸರು 

`type` - ಈ ಗುಣಲಕ್ಷಣದ ಡೇಟಾ ಪ್ರಕಾರ 

`description` - ನಿರ್ದಿಷ್ಟ ಗುಣಲಕ್ಷಣದ ವಿವರಣೆ 


**ಐಚ್ಛಿಕ**

`required` - ಫಂಕ್ಷನ್ ಕಾಲ್ ಸಂಪೂರ್ಣಗೊಳಿಸಲು ಅಗತ್ಯವಾದ ಗುಣಲಕ್ಷಣ 


### ಫಂಕ್ಷನ್ ಕರೆದೊಯ್ಯುವುದು  
ಫಂಕ್ಷನ್ ಅನ್ನು นิರ್ದಿಷ್ಟಗೊಳಿಸಿದ ನಂತರ, ನಾವು ಅದನ್ನು Chat Completion API‌ಗೆ ಕರೆ ಮಾಡುವಲ್ಲಿ ಸೇರಿಸಬೇಕು. ನಾವು ಇದು `functions` ಅನ್ನು ವಿನಂತಿಯಲ್ಲಿ ಸೇರಿಸುವ ಮೂಲಕ ಮಾಡುತ್ತೇವೆ. ಈ ಸಂದರ್ಭದಲ್ಲಿ `functions=functions` ಆಗಿದೆ.  

`function_call` ಅನ್ನು `auto` ಗೆ ಹೊಂದಿಸುವ ಆಯ್ಕೆಯೂ ಇದೆ. ಇದರ ಅರ್ಥ ನಾವು ಹಂಚಿಕೊಂಡು ಕೊಡುವುದು ಬದಲು, ಬಳಕೆದಾರ ಸಂದೇಶ ಆಧರಿಸಿ ಯಾವ ಫಂಕ್ಷನ್ ಅನ್ನು ಕರೆ ಮಾಡಬೇಕು ಎಂಬುದನ್ನು LLM ನಿರ್ಧರಿಸಲಿದೆ.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ನೀವು ನೋಡಬಹುದು ಈ ಫಂಕ್ಷನ್ ಹೆಸರನ್ನು ಕರೆಸಲಾಗಿದೆ ಮತ್ತು ಬಳಕೆದಾರ ಸಂದೇಶದಿಂದ, LLM ಫಂಕ್ಷನ್ ಆರ್ಗ್ಯುಮೆಂಶ್‌ಗಳನ್ನು ಪೂರೈಸಲು ಡೇಟಾವನ್ನು ಕಂಡುಹಿಡಿದುಕೊಂಡಿತು. 


## 3.ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ಫಂಕ್ಷನ್ ಅನ್ನು ಕರೆ ಮಾಡುವುದನ್ನು ಏಕೀಕರಣಗೊಳಿಸುವುದು. 


LLM ನಿಂದ ಸ್ವರೂಪಿತ ಪ್ರತಿಕ್ರಿಯೆಯನ್ನು ಪರೀಕ್ಷಿಸಿದ ನಂತರ, ಈಗ ನಾವು ಇದನ್ನು ಅಪ್ಲಿಕೇಶನ್‌ನಲ್ಲಿ ಏಕೀಕರಣಗೊಳಿಸಬಹುದು. 

### ಪ್ರಕ್ರಿಯೆಯನ್ನು ನಿರ್ವಹಿಸುವುದು 

ಇದನ್ನು ನಮ್ಮ ಅಪ್ಲಿಕೇಶನ್‌ಗೆ ಏಕೀಕರಣಗೊಳಿಸಲು, ನಾವು ಕೆಳಗಿನ ಹಂತಗಳನ್ನು ಅನುಸರಿಸೋಣ: 

ಮೊದಲಿಗೆ, Open AI ಸೇವೆಗಳನ್ನು ಕರೆಯಿರಿ ಮತ್ತು ಸಂದೇಶವನ್ನು `response_message` ಎಂಬ 변수ಗಳಲ್ಲಿ ಸಂಗ್ರಹಿಸೋಣ. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ಈಗ ನಾವು ಕೋರ್ಸುಗಳ ಪಟ್ಟಿಯನ್ನು ಪಡೆಯಲು Microsoft Learn API ಅನ್ನು ಕರೆ ಮಾಡಲಿರುವ ಕಾರ್ಯವನ್ನು ವ್ಯಾಖ್ಯಾನಿಸೋಣ: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ಉತ್ತಮ ಅಭ್ಯಾಸವಾಗಿ, ನಾವು ನಂತರ ಮಾದರಿ ಕಾರ್ಯವನ್ನು ಕರೆಸಲು ಇಚ್ಛಿಸುವುದೇನೋ ನೋಡಲಿದ್ದೇವೆ. ಅದರ ನಂತರ, ನಾವು ಲಭ್ಯವಿರುವ ಕಾರ್ಯಗಳಲ್ಲಿ ಒಂದನ್ನು ರಚಿಸಿ, ಕರೆಮಾಡಲಾಗುತ್ತಿರುವ ಕಾರ್ಯಕ್ಕೆ ಹೊಂದಿಸುವೆವು.
ನಂತರ, ನಾವು ಕಾರ್ಯದ ಆರ್ಗ್ಯುಮೆಂಟ್‌ಗಳನ್ನು ತೆಗೆದುಕೊಂಡು ಅವುಗಳನ್ನು LLMನ ಆರ್ಗ್ಯುಮೆಂಟ್‌ಗಳಿಗೆ ನಕ್ಷೆಮಾಡುತ್ತೇವೆ.

ಕೊನೆಗೆ, ನಾವು ಕಾರ್ಯದ ಕರೆ ಸಂದೇಶ ಮತ್ತು `search_courses` ಸಂದೇಶದಿಂದ ಹಿಂತಿರುಗಿದ ಮೌಲ್ಯಗಳನ್ನು ಸೇರಿಸುತ್ತೇವೆ. ಇದರಿಂದ LLMಗೆ
ಸ್ವಾಭಾವಿಕ ಭಾಷೆಯನ್ನು ಬಳಸಿ ಬಳಕೆದಾರರಿಗೆ ಪ್ರತಿಕ್ರಿಯಿಸುವುದಕ್ಕೆ ಬೇಕಾದ ಎಲ್ಲಾ ಮಾಹಿತಿ ಸಿಗುತ್ತದೆ.


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ಈಗ ನಾವು ನವೀಕೃತ ಸಂದೇಶವನ್ನು LLM ಗೆ ಕಳುಹಿಸುತ್ತೇವೆ ಆದ್ದರಿಂದ ನಾವು API JSON ರೂಪದಲ್ಲಿ ಇದ್ದ ಉತ್ತರದ بدر به ಅದಕೂರಿಗೆ ಸ್ವಾಭಾವಿಕ ಭಾಷೆಯ ಉತ್ತರವನ್ನು ಪಡೆಯಬಹುದು.


```python
print("Messages in next request:")
print(messages)
print()

second_response = client.responses.create(
    input=messages,
    model=deployment,
    tools=functions,
    tool_choice="auto",
    temperature=0,
    store=False,
        )  # get a new response from the model where it can see the function response


print(second_response.output_text)
```
