# ပုံဖန်တီးခြင်းလျှင််းကုဒ်များ တည်ဆောက်ခြင်း (OpenAI)

LLM များမှာ စာသားဖန်တီးခြင်းသာမက ပုံဖန်တီးခြင်းလည်း ပြုလုပ်နိုင်ပါတယ်။ စာသားဖော်ပြချက်များမှ ပုံများ ဖန်တီးနိုင်သည်။ ပုံများသည် MedTech၊ ဆောက်လုပ်ရေး၊ ခရီးသွားလုပ်ငန်း၊ ဂိမ်းထုတ်လုပ်ရေး၊ မားကက်တင်းနဲ့ အခြားသော မျိုးစုံလုပ်ငန်းများမှာ အသုံးဝင်သည်။ ဒီသင်ခန်းစာမှာ OpenAI ပလက်ဖောင်းပေါ်ရှိ **GPT Image** မော်ဒယ်များနှင့် လုပ်ကိုင်ပါမယ်။

## သင်ယူရမည့်အချက်များ

သင်ခန်းစာအဆုံးမှာ သိရှိနိုင်မှာမှာ -

- ပုံဖန်တီးခြင်း ဆိုတာဘာလဲ၊ ဘယ်နေရာတွင် အသုံးဝင်လဲဆိုတာရှင်းပြနိုင်မယ်။
- `gpt-image` မော်ဒယ်မိသားစုနှင့် အရင် DALL·E မော်ဒယ်များနှင့် မည်သို့ကွဲပြားနေသလဲကို နားလည်နိုင်မယ်။
- ပုံဖန်တီးခြင်းကွန်ရက်တစ်ခုကို တည်ဆောက်ပြီး ပုံများကို ပြင်ဆင်နိုင်မယ်။

## ပုံဖန်တီးခြင်းဆိုတာဘာလဲ?

ပုံဖန်တီးမော်ဒယ်များသည် စာသားဖော်ပြချက်မှ ပုံများဖန်တီးပေးသည်။ `gpt-image` ကဲ့သို့ခေတ်မှီ မော်ဒယ်များသည် လေ့ကျင့်မှုအတွင်း စာသားနှင့် ပုံများအကြား ဆက်စပ်မှုကို သင်ယူပြီး၊ မကျေမှု သံလိုက်များကို ဖော်ပြချက်နှင့် ကိုက်ညီသော ပုံတစ်ပုံအဖြစ် တည်ဆောက်ပေးသည်။

ပုံဖန်တီးမော်ဒယ်များသည် အထင်အရှားရှိသော မိသားစု နှစ်ခုရှိသည် -

- **`gpt-image` (OpenAI)** — ဒီသင်ခန်းစာအတွက် အသုံးပြုနေသော လက်ရှိ မော်ဒယ်။ စာသားမှ ပုံဖန်တီးခြင်းနှင့် ပုံပြင်ဆင်ခြင်း (မျက်နှာဖုံးဖြင့်ျပင်ဆင်ခြင်း) ကို ထောက်ပံ့သည်။
- **Midjourney** — ၎င်း၏ဝန်ဆောင်မှုနှင့် Discord စနစ်အခြေပြု ထုတ်လုပ်မှုဖြင့် လူကြိုက်များသော အခြား တတိယပုဂ္ဂိုလ် မော်ဒယ်တစ်ခု။

> အရင်က OpenAI ပုံဖန်တီးမော်ဒယ်များ — **DALL·E 2** နှင့် **DALL·E 3** — သည် အဆင့်မောင်းဟောင်းများ ဖြစ်သည်။ DALL·E 3 သည် သစ်တင်တပ်ဆင်ဖို့ မရရှိတော့ပါ၊ `create_variation` ကဲ့သို့သော လုပ်ဆောင်ချက်များသည် DALL·E 2 မှသာ ရှိသည်။ အသစ်တင်သော ဆော့ဖ်ဝဲများအတွက် `gpt-image` မော်ဒယ်ကို အသုံးပြုပါ။

> **အရေးကြီးချက်** - `gpt-image` မော်ဒယ်များသည် ဖန်တီးပုံကို URL အစား **base64** (`b64_json`) အဖြစ် ပြန်လည်ပေးပို့သည်။ သင့်ကုဒ်သည် base64 string ကို bytes ပြန်လည် decode နှင့် သိမ်းဆည်းရမည်ဖြစ်ပြီး၊ ပုံ URL ကို ဒေါင်းလုပ်လုပ်ရန် မရှိပါ။


## သင်၏ ပထမဆုံး ဓာတ်ပုံထုတ်လုပ်မှု အက်ပလီကေးရှင်း တည်ဆောက်ခြင်း

အဲဒါဆို ဓာတ်ပုံထုတ်လုပ်မှု အက်ပလီကေးရှင်း တည်ဆောက်ဖို့ ဘာတွေလိုပါသလဲ? သင်အောက်ပါ စာကြည့်တမ်းတွေလိုအပ်ပါတယ်။

- **python-dotenv**, သင့်ရဲ့ နှစ်သက်သော စာဝှက်ချက်များကို ကုဒ်နှင့်အတူ မပါရောက်အောင် *.env* ဖိုင်၌ သိမ်းဆည်းရန် အထူးအကြံပြုသည်။
- **openai**, OpenAI API နှင့် ဆက်သွယ်ရန် သင်အသုံးပြုမည့် စာကြည့်တိုက်။
- **pillow**, Python တွင် ဓာတ်ပုံများနှင့် ဆောင်ရွက်ရန်။


1. အောက်ပါအကြောင်းအရာပါ *.env* ဖိုင် တစ်ခု ဖန်တီးပါ-

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. အထက်ပါစာကြည့်တိုက်များကို *requirements.txt* ဆိုတဲ့ဖိုင်ထဲမှာ အောက်ပါအတိုင်းစုစည်းပါ။

    ```text
    python-dotenv
    openai
    pillow
    ```


1. နောက်တစ်ဆင့်မှာ virtual environment ဖန်တီးပြီး စာကြည့်တိုက်တွေကို ထည့်သွင်းပါ။


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Windows အတွက် virtual environment ကို ဖန်တီးပြီး အက်တိဗိတ်လုပ်ရန် အောက်ပါ command များကို အသုံးပြုပါ။

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ဟုခေါ်သော ဖိုင်ထဲ၌ အောက်ပါ code ကို ထည့်ပါ။

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv ကို आयात करें
    dotenv.load_dotenv()

    # OpenAI အရာဝတ္တ object ကို ဖန်တီးပါ (သင်၏ .env မှ OPENAI_API_KEY ကို ဖတ်သည်)
    client = openai.OpenAI()


    try:
        # Image generation API ကို အသုံးပြု၍ ပုံတစ်ပုံ ဖန်တီးပါ
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင့် prompt စာသားကို ဒီမှာ ထည့်ပါ
            size='1024x1024',
            n=1
        )
        # သိုလှောင်ထားသောပုံအတွက် ဖိုင်လ် ဒိုင်းရက်ထရီကို သတ်မှတ်ပါ
        image_dir = os.path.join(os.curdir, 'images')

        # ဒိုင်းရက်ထရီ မရှိပါက ဖန်တီးပါ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ပုံလမ်းကြောင်း အစပြုပါ (ဖိုင်အမျိုးအစားသည် png ဖြစ်သင့်သည်ကို သတိပြုပါ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image မော်ဒယ်များသည် ပုံကို base64 (b64_json) အနေနှင့် ပြန်ပေးပို့သည်၊ URL မဟုတ်ပါ
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ပုံကို ပုံလှည့်သူ အပြည့်အဝ အသုံးပြုသူတွင် ပြသပါ
        image = Image.open(image_path)
        image.show()

    # exception များကို ဖမ်းယူပါ
    except openai.BadRequestError as err:
        print(err)

    ```

ဒီ code ကို ရှင်းပြမယ်။

- ပထမဆုံးအနေဖြင့် အလိုရှိသော libraries များဖြစ်သော OpenAI library, dotenv library, base64 module နှင့် Pillow library ကို import လုပ်သည်။

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- ထို့နောက် client ကို ဖန်တီးသည်၊ client သည် API key ကို သင့် ``.env`` မှ ဖတ်ယူသည်။

    ```python
    # OpenAI အရာဝတ္ထုကို ဖန်တီးပါ။
    client = openai.OpenAI()
    ```

- နောက်တစ်ခုမှာ ပုံကို generate လုပ်သည်။

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင့် prompt စာသားကို ဒီမှာ ထည့်ပါ
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` မော်ဒယ်များသည် ပုံကို `data[0].b64_json` တွင် **base64** string အဖြစ် ပြန်ပေးသွားသည်။ အဲဒီ string ကို bytes အဖြစ် decode လုပ်ပြီး ဖိုင်ထဲ ရေးသားသည် — ဒေါင်းလုဒ်လုပ်ရန် URL မရှိပါ။

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- နောက်ဆုံးတွင် ပုံကို ဖွင့်ပြီး ပုံပရိသတ် viewer ဖြင့် ပြသသည်။

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ပုံ generate လုပ်ခြင်းနှင့် ပိုမိုနက်ရှိုင်းသော အသေးစိတ်

`images.generate` ၏ parameters များကို ကြည့်ကြမယ်။

- **model** သည် အသုံးပြုမည့် ပုံမော်ဒယ်ဖြစ်သည် (ဥပမာ `gpt-image-1`)။
- **prompt** သည် ပုံ generate လုပ်ရန် အသုံးပြုမည့် စာသား prompt ဖြစ်သည်။ ဒီမှာမှာ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ဟူ၍ ရေးထားသည်။
- **size** သည် generate လုပ်မည့် ပုံ၏ အရွယ်အစားဖြစ်သည် (ဥပမာ `1024x1024`, `1536x1024`, `1024x1536` သို့မဟုတ် `"auto"` )။
- **n** သည် generate လုပ်မည့် ပုံအရေအတွက်ဖြစ်သည်။ ဒီနောက််မှာ တစ်ပုံ generate လုပ်သည်။

> ပုံမော်ဒယ်များတွင် `temperature` parameter မရှိပါ — ၎င်းသည် စာသား generate ထိန်းချုပ်မှုဖြစ်သည်။ အမျိုးမျိုးဆန်းသော ပုံများရရန် API ကို ထပ်ခေါ်ပါ၊ အမျိုးမျိုးနည်းသည့် ပုံများအတွက် prompt ကို ပိုသေချာစေရန် လုပ်ပါ။

## ပုံ generate လုပ်မှု၏ ထပ်ဆောင်းစွမ်းဆောင်ရည်များ

ပုံ တစ်ပုံ generate လုပ်နည်းကို Python စာကြောင်းအနည်းငယ်ဖြင့် ကြည့်ရှုခဲ့ပြီး ဖြစ်သည်။ `gpt-image` မော်ဒယ်များသည် ပုံရှိပြီးသားကို **ပြင်ဆင်** နိုင်ပါသည်။ ပုံတစ်ပုံ၊ မသေချာသော အပိုင်းတစ်ခုအတွက် အမှတ်အသားထားသော **mask** (ပြင်ရန် လမ်းညွှန်ချက်) နှင့် prompt တန်းဖြင့် ပုံ၏ အပိုင်းတစ်စိတ်တစ်ပိုင်းကို ပြောင်းလဲနိုင်သည် — ဥပမာ တစ်ကောင်သော ကျားမကြွပ်ပုံပေါ်တွင် ဦးထုပ်တစ်ခု ထည့်နိုင်သည်။

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ပြင်ဆင်မှုများကို base64 အဖြစ်လည်း ပြန်ပေးပို့သည်။
edited_image = base64.b64decode(response.data[0].b64_json)
```

အခြေခံပုံမှာ ကြွပ်တစ်ကောင်သာပါရှိပြီး နောက်ဆုံးပုံတွင် ကြွပ်ပေါ် ဦးထုပ် ပါရှိသည်။
