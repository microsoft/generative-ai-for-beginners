## DALL-E နှင့် Midjourney ဆိုတာဘာလဲ?

### DALL-E

DALL-E ဖြင့် စတင်ကြမယ်၊ ၎င်းသည် စာသားဖော်ပြချက်များမှ ပုံများ ဖန်တီးသည့် Generative AI မော်ဒယ်ဖြစ်သည်။

> [DALL-E သည် မော်ဒယ်နှစ်ခု ပေါင်းစပ်ထားခြင်း ဖြစ်သည်၊ CLIP နှင့် diffused attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)။  

- **CLIP** သည် ပုံများနှင့် စာသားများမှ ဒေတာကို နံပါတ်စနစ်ဖြင့် ကိုယ်စားပြုထားသည့် embedding များ ဖန်တီးပေးသော မော်ဒယ် ဖြစ်သည်။  

- **Diffused attention** သည် embedding များမှ ပုံများ ဖန်တီးသော မော်ဒယ်ဖြစ်သည်။ DALL-E ကို ပုံနှင့် စာသား ဒေတာစုစည်းမှုဖြင့် သင်ကြားပြီး စာသားဖော်ပြချက်အတိုင်း ပုံများ ဖန်တီးနိုင်သည်။ ဥပမာ - ကွင်းထဲ နေတက်တဲ့ ကြောင်ပုံ၊ မိုဟော့ဆံတစ်မျိုးကိုင်သည့် ခွေးပုံ စသည်ဖြစ်သည်။

## DALL-E နှင့် Midjourney မော်ဒယ်တို့ လုပ်ဆောင်ပုံ

အရင်ဆုံး [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst)။ DALL-E သည် Transformer ပုံစံအခြေခံထားသော Generative AI မော်ဒယ်ဖြစ်ပြီး *autoregressive transformer* ဖြစ်သည်။  

*autoregressive transformer* သည် စာသားဖော်ပြချက်မှ ပုံကို ဘယ်လို ထုတ်လုပ်မည်ကို သတ်မှတ်ပေးပြီး ခြေပစ်ခြား Pixel တစ်ခုစီစီ ဖန်တီး သွားကာနောက်တစ်ဖက် Pixel ကို ထပ်မံ ဖန်တီးသွားသည်။ အနက်ရှိုင်း အလွှာများဖြင့် လွန်ကျူး ဆောင်ရွက်ပြီး ပုံပြီးမြောက်သည်။  

ဤလုပ်ငန်းစဉ်ဖြင့် DALL-E သည် ပုံဖန်တီးရာတွင် attribute များ၊ objects များ၊ အင်္ဂါရပ်များ နှင့် အခြား အချက်အလက်များကို ထိန်းချုပ်နိုင်သည်။ သို့သော် DALL-E 2 နှင့် 3 သည် ပိုမို ထိန်းချုပ်မှုပေးနိုင်သည်။


## သင့်ရဲ့ပထမဆုံး ပုံဖန်တီးမှု အက်ပ်လီကေးရှင်းကို တည်ဆောက်ခြင်း

အဲဒါဆို ပုံဖန်တီးမှုအက်ပ်လီကေးရှင်းတစ်ခုကို တည်ဆောက်ဖို့ ဘာတွေလိုပါသလဲ? အောက်ပါပေါင်းသည့်စာကြည့်တိုက်တွေလိုအပ်ပါတယ်။

- **python-dotenv**, သင့်ရဲ့ လျှို့ဝှက်ချက်တွေကို ကုဒ်ကနေ သီးသန့် *.env* ဖိုင်ထဲမှာ ထိန်းသိမ်းဖို့ အဆိုပါစာကြည့်တိုက်ကို အထူးအကြံပြုပါတယ်။
- **openai**, OpenAI API နဲ့ ဆက်သွယ်ရန်အသုံးပြုမည့် စာကြည့်တိုက်ဖြစ်သည်။
- **pillow**, Python တွင် ပုံများနှင့် အလုပ်လုပ်ရန်။
- **requests**, HTTP တောင်းဆိုမှုများ ထုတ်လုပ်နိုင်ရန် ကူညီပေးသည်။


1. အောက်ပါ ပါဝင်သည့် ဖိုင် *.env* တစ်ဖိုင်ကို တည်ဆောက်ပါ။

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. အထက်ဖော်ပြထားသည့် လိုင်ဘရယ်ရီများကို *requirements.txt* ဆိုသည့် ဖိုင်တစ်ခုတွင် အောက်ပါအတိုင်း စုပုံပါ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. နောက်တစ်ဆင့်တွင် virtual environment ကိုဖန်တီးပြီး လိုင်ဘရယ်ရီများကို 설치 (install) ပြုလုပ်ပါ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows အတွက် သင့်ရဲ့ virtual environment ကို ဖန်တီးပြီး အလုပ်လုပ်အောင်လုပ်ဖို့အတွက် အောက်ပါ command များကို အသုံးပြုပါ။

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ဆိုတဲ့ဖိုင်ထဲမှာ အောက်ပါကုဒ်ကို ထည့်ပါ။

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ကိုယူသွင်းပါ
    dotenv.load_dotenv()

    # OpenAI အရာဝတ္တုပြုလုပ်ပါ
    client = openai.OpenAI()


    try:
        # ပုံဖန်တီးမှု API ကိုအသုံးပြု၍ ပုံတစ်ပုံဖန်တီးပါ
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # မိမိရိုက်ထည့်လိုသော စာသားကိုဤတွင်ထည့်ပါ
            size='1024x1024',
            n=1
        )
        # သိမ်းဆည်းထားသည့် ပုံအတွက် ဖိုင်လမ်းကြောင်း သတ်မှတ်ပါ
        image_dir = os.path.join(os.curdir, 'images')

        # ဖိုင်လမ်းကြောင်းမရှိပါက ဖန်တီးပါ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ပုံလမ်းကြောင်းကို စတင်ထားပါ (ဖိုင်အမျိုးအစားမှာ png ဖြစ်ရန် သတိပြုပါ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # ဖန်တီးထားသော ပုံကို ယူပါ
        print(generation_response)

        image_url = generation_response.data[0].url # တုံ့ပြန်မှုမှ ပုံ URL ကို ထုတ်ယူပါ
        generated_image = requests.get(image_url).content  # ပုံကိုဒေါင်းလုပ်လုပ်ပါ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ပုံကို ပုံကြည့်ကိရိယာ ပုံမှန်ဖြင့် ပြပါ
        image = Image.open(image_path)
        image.show()

    # အမှားများကို ဖမ်းဆီးပါ
    except openai.BadRequestError as err:
        print(err)

    ```


- ထို့နောက် ``.env`` များမှ API key ကို ရယူမယ့် object ကို ဖန်တီးပါတယ်။

    ```python
        # OpenAI အရာဝတ္ထုကို ဖန်တီးပါ
        client = openai.OpenAI()
    ```

- နောက်တစ်ခုတော့ ပုံရိပ်ကို generate လုပ်ပါတယ်။

    ```python
    # အင်မတန်ပုံဖန်တီးခြင်း API ကို အသုံးပြု၍ ပုံတစ်ပုံကို ဖန်တီးပါ
    generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင်၏အကြံဉာဏ်စာသားကို ဒီနေရာတွင် ထည့်ပါ
        size='1024x1024',
        n=1
    )
    ```

    အထက်ဖော်ပြတဲ့ကုဒ်က JSON object ရပါတယ်၊ ဒီ object မှာ generated image ရဲ့ URL ပါဝင်တယ်။ ဒီ URL ကိုအသုံးပြုပြီး ပုံရိပ်ကို download အဖြစ် ထည့်သိမ်းနိုင်ပါတယ်။

- နောက်ဆုံးမှာတော့ ပုံရိပ်ကို ဖွင့်ပြီး ပုံရိပ် viewer စနစ်ကို အသုံးပြုပြီး ပြပါသည်။

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### ပုံရိပ်ဖန်တီးခြင်းအကြောင်း အနက်အကြောင်း အသစ်စွာ ရှင်းလင်းခြင်း

ပုံရိပ် ဖန်တီးခြင်းအကြောင်း ပိုမိုအသေးစိတ် ရှင်းပြထားသည့်ကုဒ်ကို ကြည့်ကြပါစို့။

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင့်၏ စကားဝိုင်း စာသားကို ဒီမှာ ဖြည့်စွက်ပါ
        size='1024x1024',
        n=1
    )
```

- **prompt** ဆိုတာက ပုံရိပ် ဖန်တီးရာမှာ အသုံးပြုတဲ့ စာသား prompt ဖြစ်သည်။ ဒီမှာတော့ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" လို့ရေးထားတယ်။
- **size** က ဖန်တီးချင်တဲ့ ပုံရိပ်ရဲ့ အရွယ်အစား ဖြစ်ပါတယ်။ ဒီမှာတော့ 1024x1024 pixels ဖြစ်အောင် generate လုပ်ထားတယ်။
- **n** က ဖန်တီးမယ့် ပုံရိပ် အရေအတွက် ဖြစ်သည်။ ဒီမှာတော့ ပုံ ၂ ပုံ generate လုပ်နေပါသည်။

ပုံရိပ်များနဲ့ ဆက်စပ်ပြီး နောက်ပိုင်းပိုလျောက် နည်းလမ်းများကိုလည်း ကျွန်ုပ်တို့ ၎င်းတွင် ဖော်ပြပေးပါမယ်။

## ပုံရိပ် ဖန်တီးခြင်း၏ နောက်ထပ်လုပ်ဆောင်ချက်များ

Python မှာ လိုက်လံရေးသားသည့်အတိုင်း ဗဟိုချက်နည်းနည်းဖြင့်ပုံရိပ် တစ်ပုံဖန်တီးနိုင်ခဲ့ပြီ ဖြစ်သည်။ သို့သော် ပုံရိပ်နဲ့ ဆက်စပ်ပြီး အခြားလည်း ရနိုင်သည်။

ကိုယ်လည်း အောက်ပါအရာတွေကို လုပ်ဆောင်နိုင်ပါသည်။

- **ပြင်ဆင်ချက်များလုပ်ဆောင်ခြင်း**။ ရှိပြီးသားပုံရိပ် တစ်ပုံအတွက် mask နဲ့ prompt ပေးပြီး ပြုပြင်နိုင်ပါသည်။ ဥပမာအားဖြင့် bunny ပုံရိပ်မှာ hat တစ်ခု ထည့်လိုပါက မူလ ပုံရိပ်၊ mask (ပြင်ဆင်မယ့်နေရာကို ခွဲခြားဖော်ပြချက်), နှင့် ပြင်ဆင်စေလိုသော စာသား prompt တို့ကို ဖြည့်ပါ။

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    အခြေခံပုံမှာတော့ မြင်းတစ်ကောင်သာ ပါမယ်၊ ဒါပေမဲ့ နောက်ဆုံးပုံမှာ မြင်းပေါ်က hat ပါလာပါလိမ့်မယ်။
    
- **ပုံကွဲဖန်တီးခြင်း**။ ရှိပြီးသားပုံတစ်ပုံကိုယူပြီး အမျိုးမျိုးသောပုံကွဲများဖန်တီးရန် တောင်းဆိုနိုင်သည်။ ပုံကွဲတစ်ခု ဖန်တီးရန်အတွက် ပုံတစ်ပုံနဲ့ စာသား prompt တစ်ခုပေးပြီး အောက်ပါအတိုင်း code မှာရေးပါ။

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
