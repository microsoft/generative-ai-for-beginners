## DALL-E နှင့် Midjourney ဆိုသည်မှာ အဘယ်နည်း?

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) နှင့် [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) သည် လူကြိုက်အများဆုံး ပုံဖန်တီးမှု မော်ဒယ် ၂ မျိုးဖြစ်ပြီး ပရော့မ့်များ အသုံးပြု၍ ပုံများ ဖန်တီးနိုင်စေသည်။

### DALL-E

DALL-E သည် စာသားဖော်ပြချက်များမှ ပုံများ ဖန်တီးပေးသည့် Generative AI မော်ဒယ်တစ်ခု ဖြစ်သည်။

> [DALL-E သည် CLIP နှင့် diffused attention မော်ဒယ် နှစ်ခု ပေါင်းစပ်ထားခြင်းဖြစ်သည်](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)။

- **CLIP** သည် ပုံများ နှင့် စာသားများမှ ဒေတာ၏ နံပါတ်ပြု အမျိုးအစားများ ဖြစ်သော embedding များ ဖန်တီးပေးသည့် မော်ဒယ်ဖြစ်သည်။

- **Diffused attention** သည် embedding များမှ ပုံများ ဖန်တီးပေးသည့် မော်ဒယ်ဖြစ်သည်။ DALL-E သည် ပုံများနှင့် စာသားများ အချက်အလက် စုစည်းထားသည့် ဒေတာ အစုအဝေးဖြင့် လေ့ကျင့်ပြီး စာသား ဖော်ပြချက်များမှ ပုံများ ဖန်တီးနိုင်သည်။ ဥပမာအားဖြင့် DALL-E သည် ဖိနပ်တပ်ထားသည့် ကြောင်ပုံ သို့မဟုတ် မိုဟော့(နောက်ကျောရိုး)ရှိသော ခွေးပုံ စသည့် ပုံများ ဖန်တီးနိုင်ပါသည်။

### Midjourney
 
Midjourney သည် DALL-E နှင့် ဆင်တူပြီး စာသားပရော့မ့်များမှ ပုံများ ဖန်တီးပေးပါသည်။ Midjourney သည် စာသားပရော့မ့်တစ်ခု "ဖိနပ်တပ်ထားသည့် ကြောင်" သို့မဟုတ် "မိုဟော့ရှိသော ခွေး" ကဲ့သို့ အသုံးပြု၍ ပုံများ ဖန်တီးနိုင်သည်။

## DALL-E နှင့် Midjourney မော်ဒယ်များ မည်သို့လည်ပတ်ကြသနည်း

ပထမဆုံး [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) ကို ကြည့်ပါမယ်။ DALL-E သည် transformer အခြေခံပြီး *autoregressive transformer* အမျိုးအစားဖြစ်သည်။

*autoregressive transformer* သည် စာသားဖော်ပြချက်မှ ပုံဖန်တီးရာတွင် တစ်ခုချင်း pixel ကို အဆင့်လိုက် ဖန်တီးသွားပြီး နောက်တစ်ခုကို ဖန်တီးရာတွင် ယခင် ဖန်တီးထားသော pixel များကို အသုံးပြုသည့် နည်းပညာဖြစ်သည်။ နယူးရယ်ကွန်ရက် အလွှာများထံ ဖြတ်ကူးသွားပြီးနောက် ပုံပြီးစီးပါသည်။

ဤနည်းဖြင့် DALL-E သည် ပုံတွင် ပါဝင်သော attribute များ, object များ, ချက်များ နှင့် အခြား စိတ်ကြိုက် ထိန်းချုပ်နိုင်သည်။ သို့သော် DALL-E 2 နှင့် 3 သည် ပိုပြီး ပုံထုတ်လုပ်မှုတွင် ထိန်းချုပ်မှုရှိသည်။


## သင်၏ ပထမဆုံး ပုံ စုဆောင်းရေး အက်ပလီကေးရှင်းကို ဆောက်ခြင်း

အဲဒါဆို ပုံ စုဆောင်းရေး အက်ပလီကေးရှင်းတစ်ခု ဆောက်ရန် ဘာများ လိုအပ်ပါသလဲ? အောက်ပါ စာကြည့်တိုက်များ လိုအပ်သည်။

- **python-dotenv**, သင့်ရဲ႕ လျှို့ဝှက်ချက်များကို ကုဒ်နဲ့ကွာရှင်းရန် *.env* ဖိုင်ထဲမှာ သိမ်းဆည်းရန်အတွက် ဒီစာကြည့်တိုက်ကို အကြံပြုပါတယ်။
- **openai**, ဒီစာကြည့်တိုက်ကို OpenAI API နှင့် ဆက်သွယ်ရန် အသုံးပြုမည်ဖြစ်သည်။
- **pillow**, Python တွင် ပုံများနှင့် လုပ်ဆောင်ရန်။
- **requests**, HTTP မေးမြန်းမှုများ ပြုလုပ်ရန် အကူအညီဖြစ်စေသည်။

မပြုလုပ်မှီဖြစ်ပါက [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) စာမျက်နှာအတိုင်း လိုက်နာကာ Azure OpenAI resource နှင့် မော်ဒယ် တည်ဆောက်ပါ။ မော်ဒယ်အဖြစ် **gpt-image-1** ကို ရွေးချယ်ပါ (လက်ရှိမြင့်မားသော Azure OpenAI ပုံမော်ဒယ်ဖြစ်ပြီး DALL-E 3 သည် အလားအလာမရှိတော့သော နှင့် အခြား အသစ်တင်ဆက်မှုများအတွက် မရရှိတော့ပါ)။

1. *.env* ဆိုသော ဖိုင် ကို အောက်ပါအတိုင်း ဖန်တီးပါ။

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    သင့် resource အတွက် Microsoft Foundry portal ၏ "Deployments" အပိုင်းတွင် ဒီအချက်အလက်များကို ရှာဖွေပါ။



1. အထက်ဖော်ပြထားသည့် စာကြည့်တိုက်များအား *requirements.txt* ဟု အမည်ပေးထားသော ဖိုင်တွင် အောက်ပါအတိုင်း စုဆောင်းပါ။

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. နောက်တစ်ဆင့်မှာ virtual environment ဖန်တီးပြီး စာကြည့်တိုက်များကို 설치 လုပ်ပါ။


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows အတွက်တော့ မိမိရဲ့ virtual environment ကို ဖန်တီးပြီး ဖွင့်ရန် အောက်ပါ command များကို အသုံးပြုနိုင်သည်။

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ဆိုတဲ့ဖိုင်ထဲမှာ အောက်ပါ code ကို ထည့်ပါ။

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv ကို ထည့်သွင်းသည်
    dotenv.load_dotenv()

    # Azure OpenAI ဝန်ဆောင်မှု အိုင်လျှောက်ကုန်စီမံကိန်းကို စီမံခန့်ခွဲသည်
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # ပုံဖန်တီးမှု API ကို အသုံးပြု၍ ပုံတစ်ပုံ ဖန်တီးပါ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင်၏ ပြည့်စုံသော စာသားကို ဒီမှာ ထည့်ပါ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # သိမ်းဆည်းထားသော ပုံ၏ ဖိုလ်ဒါ လမ်းကြောင်း သတ်မှတ်ပါ
        image_dir = os.path.join(os.curdir, 'images')
    
        # ဖိုလ်ဒါ မရှိပါက ဖန်တီးပါ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # ပုံလမ်းကြောင်း သတ်မှတ်ခြင်း (ဖိုင်အမျိုးအစားသည် png ဖြစ်ရန် သတိပြုပါ)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # ဖန်တီးထားသော ပုံကို ပြန်လည်ရယူပါ
        image_url = generation_response.data[0].url  # အဖြေမှ ပုံ URL ကို ခွဲထုတ်ပါ
        generated_image = requests.get(image_url).content  # ပုံကို ဒေါင်းလုပ်လုပ်ပါ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # ပုံအား မူရင်းပုံကြည့်ဖွင့်နိုင်စက်တွင် ပြပါ
        image = Image.open(image_path)
        image.show()
    
    # အမှားများကို ဖမ်းယူပါ
    except openai.BadRequestError as err:
        print(err)

    ```

### ပုံ generate လုပ်ခြင်းအကြောင်း အသေးစိတ်

ပုံ generate လုပ်တဲ့ code ကို အသေးစိတ် ကြည့်ကြမယ်။

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင့်မေးခွန်းစာသားကို ဤနေရာတွင် ထည့်ပါ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, ဆိုသည်မှာ ပုံ generate လုပ်ရန် အသုံးပြုသော စာသား prompt ဖြစ်သည်။ ဤနောက်ဆုံးကိစ္စတွင် "မြင်းပေါ်ရှိ သားမြင်း၊ လော်လိပ်ပေါင်းကိုင်၍ အလယ်ဧက နယ်မြေမှာ ပန်းပွင့်ကြွယ်ဝတဲ့ နေရာ" လို့ ဖော်ပြထားသည်။
- **size**, ဆိုသည်မှာ generate လုပ်မည့် ပုံ၏ အရွယ်အစား ဖြစ်သည်။ ဤနောက်ဆုံးကိစ္စတွင် 1024x1024 ပစ်ကယ် စွပ်သုံးသည်။
- **n**, ဆိုသည်မှာ generate လုပ်မည့် ပုံပေါင်း၏ အရေအတွက် ဖြစ်သည်။ ဤနောက်ဆုံးကိစ္စတွင် ပုံတစ်ပုံ ဖန်တီးသည်။

ပုံများနှင့်ဆက်စပ်ပြီး များစွာလုပ်ဆောင်နိုင်သည့် လုပ်ဆောင်ချက်များကို နောက်ပိုင်းအပိုင်းတွင် ကိုယ်တိုင် လေ့လာကြမယ်။

## ပုံ generate လုပ်ခြင်း၏ ပိုမိုသော စွမ်းဆောင်ရည်များ

Python က ဆော့ဖ်ဝဲ တစ်ခုဖြင့် ခဏတာအတွင်း ပုံ generate လုပ်နိုင်ခြင်းကို ကြည့်ရှုခဲ့ပြီး ဖြစ်သည်။ သို့သော် ပုံများနှင့် ဆက်နွယ်ရာ အခြားလုပ်ဆောင်နိုင်ချက်များလည်း ရှိသည်။

များစွာလုပ်ဆောင်နိုင်သော အရာများမှာ အောက်ပါအတိုင်းဖြစ်သည်။

- **ပြုပြင်ပြောင်းလဲမှုလုပ်ခြင်း**။ ရှိ အရင်ကပုံတစ်ပုံအပေါ် mask တစ်ခုနှင့် prompt တစ်ခု ပေးခြင်းဖြင့် ပုံကို ပြုပြင်ပြောင်းလဲနိုင်သည်။ ဥပမာ၊ ငါတို့၏သားမြင်းပုံတွင် ရှိသည့်သားမြင်းအပေါ် နွားခေါင်းစည်းတစ်ခု ထည့်နိုင်သည်။ ထိုလုပ်ငန်းစဉ်ကို ဓာတ်ပုံ၊ mask (ပြင်ဆင်ရန် နေရာ ကို ပြသသော), နှင့် ပြင်ဆင်မှာ မည်သို့လုပ်မည်ဆိုသော စာသား prompt ဖြင့် တင်ပြခြင်းဖြင့် ဆောင်ရွက်သည်။

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    အခြေခံ image တွင် သားမြင်းသာ ပါရှိမည် ဖြစ်ရာ နောက်ဆုံးအဖြစ်မှတ် image တွင် နွားခေါင်းစည်း ပါရှိမည်။
