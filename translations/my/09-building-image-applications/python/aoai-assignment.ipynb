# ပုံပေါ်ထွက်စက်ဆော့ဖ်ဝဲ အက်ပလီကေးရှင်းများတည်ဆောက်ခြင်း (Azure OpenAI)

LLM များမှာ စာသားထွက်စက်ထက်ပိုသည်။ သင်သည် စာသားဖော်ပြချက်များမှ ပုံများကိုလည်းထုတ်လုပ်နိုင်သည်။ ပုံများသည် မက်ဒ်နည်းပညာ၊ စီမံကိန်းရေးဆွဲခြင်း၊ ခရီးသွားလုပ်ငန်း၊ ဂိမ်းဖွံ့ဖြိုးတိုးတက်ရေး၊ မာကတာနဲ့ အခြားများတွင် အသုံးဝင်သည်။ ဒီသင်ခန်းစာမှာ Microsoft Foundry ပေါ်က ယနေ့ခေတ် **GPT ပုံထုတ်စက်** မော်ဒယ်များဖြင့် အလုပ်လုပ်ပါမယ်။

## သင်ယူလိုသည့်ရည်မှန်းချက်များ

ဒီသင်ခန်းစာပြီးဆုံးချိန်မှာ သင်မှာ အောက်ပါစွမ်းရည်များ ရရှိနေပါမယ်။

- ပုံထုတ်လုပ်ခြင်းသည် အဘယ်ကြောင့်ဖြစ်ပြီး ဘယ်နေရာတွင် အသုံးဝင်သည်ကို ရှင်းပြနိုင်သည်။
- `gpt-image` မော်ဒယ် မိသားစုကို နားလည်ကာ ယခင်ခေတ် DALL·E မော်ဒယ်များနှင့် မည်သို့ကွဲပြားသည်ကို သိရှိနားလည်နိုင်သည်။
- ပုံထုတ်လုပ်စက်နှင့် ပုံတည်းဖြတ်ခြင်း အက်ပလီကေးရှင်း တည်ဆောက်နိုင်သည်။

## ပုံထုတ်လုပ်ခြင်း ဆိုသည်မှာ ဘာလဲ?

ပုံထုတ်လုပ်မှု မော်ဒယ်များသည် စာသားအချက်အလက်မှ ပုံများကိုဖန်တီးသည်။ ခေတ်သစ် မော်ဒယ်များဖြစ်သော `gpt-image` က စာသားနှင့် ပုံများအကြား ဆက်နွယ်မှုကို သင်ကြားပြီးနောက် အဆင့်ဆင့် ကလပ်ဆူကြားမှ ပုံသို့ ပြောင်းလဲပေးသည်။

ပုံမော်ဒယ်များတွင် နာမည်ကြီး မိသားစု နှစ်မျိုးမှာ -

- **`gpt-image` (OpenAI)** — ဒီသင်ခန်းစာမှာ အသုံးပြုသော ခေတ်သစ် မော်ဒယ်ဖြစ်သည်။ စာသားမှ ပုံထုတ်ခြင်းနှင့် ပုံတည်းဖြတ်ခြင်း (ကွက်တိပန်လုပ်ခြင်း) ကို ထောက်ပံ့သည်။
- **Midjourney** — အားလုံးသိကြတဲ့ သုံးပတ်ဝန်းကျင် ကိုယ်ပိုင် ဝန်ဆောင်မှုနဲ့ Discord အခြေခံ စနစ်ပါဝင်သော မြင်သာကျယ်ပြန့်စွာအသုံးပြုသော မော်ဒယ်တစ်ခု။

> အဟောင်း OpenAI ပုံမော်ဒယ်များ — **DALL·E 2** နှင့် **DALL·E 3** — သည် အရင်းအမြစ်ဟောင်းများ ဖြစ်ကြသည်။ DALL·E 3 သည်အသစ်စက်စက် တပ်ဆင် yritynt များအတွက် ရရှိနိုင်ခြင်း မရှိတော့ပါ၊ `create_variation` ကဲ့သို့ အင်္ဂါရပ်များသည် DALL·E 2 တွင်သာ ရှိခဲ့သည်။ အသစ့်အက်ပလီကေးရှင်းများအတွက် `gpt-image` မော်ဒယ်များကို အသုံးပြုပါ။

Microsoft Foundry ပေါ်တွင် **`gpt-image-2`** သည် ခေတ်ပေါ်ပြီး အင်အားအပြည့်ဆုံး ပုံမော်ဒယ်ဖြစ်ပြီး ပုံမှန်ဖြစ်စေလိုက် Onလုပ်စရာဖြစ်သည်။ `gpt-image-1.5` နှင့် `gpt-image-1-mini` များလည်း မူလ အတွက်ရရှိနိုင်သည်။

> **အရေးကြီးချက်**: `gpt-image` မော်ဒယ်များသည် ထုတ်လုပ်ပုံကို **base64** (`b64_json`) အဖြစ် ပြန်လည်ထုတ်ပေးပြီး URL အဖြစ် မရှိပါ။ သင်၏ကုဒ်သည် base64 string ကို bytes သို့ စစ်ထုတ်ပြီး သိမ်းဆည်းရမည်ဖြစ်ပြီး URL ဖြင့် ဒေါင်းလုဒ်မရပါ။


## သင်၏ ပထမဆုံး ပုံဖန်တီးမှု အပလီကေးရှင်း တည်ဆောက်ခြင်း

ပုံဖန်တီးမှု အပလီကေးရှင်း တည်ဆောက်ဖို့ ဘာတွေလိုလဲ? အောက်ပါ 라이ဘရီများလိုအပ်သည်။

- **python-dotenv**, သင့်ရဲ့ လျှို့ဝှက်ချက်များကို ကုဒ်မှ ကွာရှိသော *.env* ဖိုင်တစ်ခုတွင် သိမ်းဆည်းထားရန် အထူးအကြံပြုပါသည်။
- **openai**, OpenAI API နှင့် ဆက်သွယ်ရန် သုံးမည့် 라이ဘရီဖြစ်သည်။
- **pillow**, Python တွင် ပုံများနှင့် ပတ်သက်ကာ လုပ်ဆောင်ရန်။

မပြီးသေးပါက၊ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) စာမျက်နှာရှိ ညွှန်ကြားချက်များအား လိုက်နာ၍ Microsoft Foundry resource နှင့် model တစ်ခု ဖန်တီးပါ။ **gpt-image-2** ကို မော်ဒယ်အနေဖြင့် ရွေးချယ်ပါ (နောက်ဆုံးထွက် Azure OpenAI ပုံမော်ဒယ်ဖြစ်ပြီး DALL·E သည် ယခင်နည်းဖြစ်သည်)။

1. *.env* ဖိုင်တစ်ခုကို အောက်ပါ အကြောင်းအရာ ဖြင့် ဖန်တီးပါ။

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    သင့် resource ၏ Microsoft Foundry portal တွင် "Deployments" အပိုင်းတွင် ဤအချက်အလက်ကို ရှာဖွေပါ။



1. အထက်ဖော်ပြပါစာကြည့်တိုက်များကို *requirements.txt* ဟုခေါ်သောဖိုင်ထဲတွင် အောက်ပါအတိုင်းစုစည်းပါ။

    ```text
    python-dotenv
    openai
    pillow
    ```


1. နောက်တစ်ဆင့်အဖြစ် virtual environment တခု ဖန်တီးပြီး စာကြည့်တိုက်များကို installs လုပ်ပါ။


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Windows အတွက် virtual environment ကို ဖန်တီးပြီး ဖွင့်ရန်အတွက် အောက်ပါ command များကို အသုံးပြုပါ။

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ဆိုတဲ့ ဖိုင်ထဲမှာ အောက်ပါ code ကို ထည့်ပါ။

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv ကို import ဆောင်ရွက်ပါ
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) client ကို configure ဆောင်ရွက်ပါ။
    # ပုံသဏ္ဍာန်စနစ်များသည် မကြာသေးမီ API ဗားရှင်းကို လိုအပ်သည် — သင့်ပုံစံလိုအပ်ချက်အတွက် Foundry စာရွက်ကို စစ်ဆေးပါ။
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # ပုံဖန်တီးရန် ပုံဖန်တီးမှု API ကို အသုံးပြုပါ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # သင်၏ prompt စာသားကို ဤနေရာတွင်ရိုက်ထည့်ပါ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # ဥပမာ - "gpt-image-2"
        )
        # သိမ်းဆည်းထားသောပုံဖိုင် ဖိုင်လမ်းညွှန်ကို သတ်မှတ်ပါ
        image_dir = os.path.join(os.curdir, 'images')

        # ဖိုင်လမ်းညွှန်မရှိပါက တည်ဆောက်ပါ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ပုံဖိုင်လမ်း (png ဖိုင်မျိုးဖြစ်ရမည်) ကို စတင်သတ်မှတ်ပါ
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image မော်ဒယ်များသည် ပုံကို URL မဟုတ်ဘဲ base64 (b64_json) အဖြစ် ပြန်လည်ပေးဆောင်သည်
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ပုံကို မူရင်းပုံကြည့်ကိရိယာတွင် ပြသပါ
        image = Image.open(image_path)
        image.show()

    # အမှားများကို ဖမ်းဆီးပါ
    except BadRequestError as err:
        print(err)

    ```

ဒီ code ကို ရှင်းပြကြပါစို့။

- မူလအနေနဲ့၊ OpenAI library, dotenv library, base64 module နဲ့ Pillow library ကို ထည့်ယူထားတယ်။

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- နောက်တစ်ချက် အနေနဲ့ *.env* ဖိုင်ထဲက environment variables ကို load လုပ်တယ်။

    ```python
    # dotenv ကို သွင်းယူပါသည်
    dotenv.load_dotenv()
    ```

- ထို့နောက် Azure OpenAI (Microsoft Foundry) client ကို configure လုပ်ကြတယ်။

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- နောက်တစ်ဆင့်မှာ image ကို generate လုပ်တယ်။

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ကၽြန္ေတာ့္ရဲ႕ စကားဝိုင္းစာသားကို ဒီမွာ ထည့်ပါ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` models သည် image ကို `data[0].b64_json` ထဲမှာ **base64** string အနေဖြင့် ပေးလှမ်းပြီး၊ အဲဒီ string ကို bytes အဖြစ် decode လုပ်၍ ဖိုင်ထဲသို့ရေးသားသည် — ဒါကြောင့် download ရရန် URL မရှိပါ။

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- နောက်ဆုံးတွင် image ကို ဖွင့်၍ ပုံမှန် image viewer ဖြင့် ပြသသည်။

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ပုံရိပ်ကို ဖန်တီးခြင်းနှင့် ပုံစံအသေးစိတ်များ

`images.generate` ၏ parameters များကို ကြည့်ရှူကြပါစို့။

- **prompt** သည် image ဖန်တီးရန် အသုံးပြုသော စာသား prompt ဖြစ်သည်။ ဒီနေရာမှာ “Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils” ဖြစ်သည်။
- **size** သည် ဖန်တီးမည့် image ၏ အရွယ်အစား ဖြစ်သည် (ဥပမာ `1024x1024`, `1536x1024`, `1024x1536`, သို့မဟုတ် `"auto"`)။
- **n** သည် ဖန်တီးမည့် image အရေအတွက် ဖြစ်သည်။ ဒီနေရာတွင် တစ်ခု ဖန်တီးသည်။
- **model** သည် သင့်ရဲ့ image model deployment အမည် ဖြစ်သည် (ဥပမာ `gpt-image-2`)။

> Image model များတွင် `temperature` parameter မပါဝင်ပါ — ဒါဟာ စာသားဖန်တီးမှု ထိန်းချုပ်မှု ဖြစ်သည်။ အမျိုးမျိုးရှိစေလိုပါက API ကို ထပ်ခေါ်ပါ; ကွဲပြားမှုကို လျော့ချလိုပါက prompt ကို ပိုမို သေချာနားလည်စေပါ။

## ပုံရိပ်ဖန်တီးမှု၏ ထပ်ဆောင်း စွမ်းဆောင်ရည်များ

Python အတော်လေးပဲသုံးပြီး ပုံရိပ်ကို ဘယ်လို generate လုပ်ရမယ်ဆိုတာ တွေ့မြင်ခဲ့ပါပြီ။ `gpt-image` models တွေကရှိတဲ့ function တွေထဲမှာ ရှိပြီးသား ပုံတစ်ပုံကို **တည်းဖြတ်** လုပ်နိုင်ပါသည်။ ပုံတစ်ပုံနဲ့ အကွက်အမှတ်တစ်ခု (mask) ដែល ပြင်ဆင်ရမယ့်နေရာကို သတ်မှတ်ပေးပြီး prompt ကနေပဲ ပုံနဲ့ ဆန်းစစ်ပြီး အပိုင်းတစ်ခုကို ပြောင်းလဲနိုင်ပါတယ်။ ဥပမာ ahaan bunny ရဲ့ ခေါင်းစည်း တစ်ခု ထည့်ရမှာ ဖြစ်ပါတယ်။

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# တည်းဖြတ်ချက်များကို base64 အဖြစ်လည်း ပြန်ပေးပို့သည်
edited_image = base64.b64decode(response.data[0].b64_json)
```

အခြေခံ image မှာ ယောင်လေးပဲ ပါရှိပြီး နောက်ဆုံး image မှာ ယောင်လေး့အပေါ် ခေါင်းစည်းပါတော့တယ်။
