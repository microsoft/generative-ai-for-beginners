## DALL-E နဲ့ Midjourney ဆိုတာဘာလဲ?

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) နဲ့ [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) ဟာ လူကြိုက်အများဆုံး ပုံထုတ်လုပ်ရေးမော်ဒယ်နှစ်ခုဖြစ်ပြီး prompt တွေသုံးပြီး ပုံထုတ်လုပ်နိုင်ပါတယ်။

### DALL-E

DALL-E က စာသားဖော်ပြချက်တွေကနေ ပုံထုတ်လုပ်နိုင်တဲ့ Generative AI မော်ဒယ်တစ်ခုပါ။

> [DALL-E ဟာ CLIP နဲ့ diffused attention ဆိုတဲ့ မော်ဒယ်နှစ်ခုပေါင်းစပ်ထားတာပါ](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)။

- **CLIP** က ပုံနဲ့စာသားတွေကနေ embedding တွေ (ဒေတာကို ဂဏန်းနဲ့ဖော်ပြထားတဲ့အရာ) ထုတ်ပေးတဲ့မော်ဒယ်ပါ။

- **Diffused attention** က embedding တွေကနေ ပုံထုတ်ပေးတဲ့မော်ဒယ်ပါ။ DALL-E ကို ပုံနဲ့စာသား dataset တွေသုံးပြီး သင်ကြားထားလို့ စာသားဖော်ပြချက်ကနေ ပုံထုတ်နိုင်ပါတယ်။ ဥပမာ DALL-E ကို "တစ်ကောင်က ဦးထုပ်တပ်ထားတဲ့ကြောင်"၊ "မိုဟော့ခေါင်းပုံစံရှိတဲ့ခွေး" စတာတွေထုတ်နိုင်ပါတယ်။

### Midjourney

Midjourney က DALL-E နဲ့ တူတဲ့နည်းလမ်းနဲ့ အလုပ်လုပ်ပါတယ်။ စာသား prompt တွေသုံးပြီး ပုံထုတ်နိုင်ပါတယ်။ ဥပမာ “တစ်ကောင်က ဦးထုပ်တပ်ထားတဲ့ကြောင်”၊ “မိုဟော့ခေါင်းပုံစံရှိတဲ့ခွေး” စတာတွေထုတ်နိုင်ပါတယ်။

## DALL-E နဲ့ Midjourney ဘယ်လိုအလုပ်လုပ်သလဲ

ပထမဆုံး [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) ကို ကြည့်ကြပါစို့။ DALL-E ဟာ transformer architecture ကိုအခြေခံထားတဲ့ Generative AI မော်ဒယ်တစ်ခုဖြစ်ပြီး *autoregressive transformer* ပါဝင်ပါတယ်။

*Autoregressive transformer* ဆိုတာက မော်ဒယ်တစ်ခုက စာသားဖော်ပြချက်ကနေ ပုံထုတ်လုပ်တဲ့နည်းလမ်းကို သတ်မှတ်ပေးတာပါ။ တစ်ပစ်ဆင့်ချင်းစီ ပုံထုတ်ပြီး neural network layer အများကြီးကို ဖြတ်သွားရင်း ပုံတစ်ပုံလုံး ပြီးစီးအောင်လုပ်ပါတယ်။

ဒီနည်းလမ်းနဲ့ DALL-E က ပုံထုတ်တဲ့ attributes, objects, characteristics စတာတွေကို ထိန်းချုပ်နိုင်ပါတယ်။ ဒါပေမယ့် DALL-E 2 နဲ့ 3 တွေက ပိုပြီး generated image ကို ထိန်းချုပ်နိုင်ပါတယ်။


## သင့်ရဲ့ ပထမဆုံး ပုံထုတ်လုပ်ရေး အက်ပ်လီကေးရှင်း တည်ဆောက်ခြင်း

ပုံထုတ်လုပ်ရေး အက်ပ်လီကေးရှင်း တစ်ခု တည်ဆောက်ဖို့ ဘာတွေ လိုအပ်ပါသလဲ? အောက်ပါ လိုက်ဘရရီတွေ လိုအပ်ပါတယ်-

- **python-dotenv** ကို သုံးဖို့ အထူးအကြံပြုပါတယ်။ ဒီလိုက်ဘရရီက သင့်ရဲ့ လျှို့ဝှက်ချက်တွေကို *.env* ဖိုင်ထဲမှာ ကုဒ်နဲ့ ခွဲထားနိုင်အောင် ကူညီပေးပါတယ်။
- **openai** ကို သုံးပြီး OpenAI API နဲ့ ဆက်သွယ်ဖို့ လိုအပ်ပါတယ်။
- **pillow** ကို Python မှာ ပုံတွေနဲ့ အလုပ်လုပ်ဖို့ သုံးပါတယ်။
- **requests** က HTTP request တွေ ပို့ဖို့ ကူညီပေးပါတယ်။

1. *.env* ဆိုတဲ့ ဖိုင်တစ်ခု ဖန်တီးပြီး အောက်ပါအတိုင်း ထည့်ပါ-

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    ```

    ဒီအချက်အလက်တွေကို Azure Portal မှာ သင့်ရဲ့ resource အတွက် "Keys and Endpoint" ဆိုတဲ့အပိုင်းထဲမှာ ရှာနိုင်ပါတယ်။


1. အထက်ပါလိုက်ဘရရီတွေကို *requirements.txt* ဆိုတဲ့ဖိုင်ထဲမှာ ဒီလိုရေးထည့်ပါ။

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. နောက်တစ်ဆင့်၊ virtual environment တစ်ခုဖန်တီးပြီး လိုအပ်တဲ့လိုက်ဘရရီတွေကို install လုပ်ပါ။


````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
````

> [!NOTE]
> Windows အတွက် သင့် virtual environment ကို ဖန်တီးပြီး အသုံးပြုရန် အောက်ပါ command များကို သုံးပါ။

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ```

1. *app.py* ဆိုတဲ့ ဖိုင်ထဲမှာ အောက်ပါ code ကို ထည့်ပါ။

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # import dotenv
    dotenv.load_dotenv()
    
    # Get endpoint and key from environment variables
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # Assign the API version (DALL-E is currently supported for the 2023-06-01-preview API version only)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # Create an image by using the image generation API
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # Set the directory for the stored image
        image_dir = os.path.join(os.curdir, 'images')
    
        # If the directory doesn't exist, create it
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # Initialize the image path (note the filetype should be png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # Retrieve the generated image
        image_url = generation_response["data"][0]["url"]  # extract image URL from response
        generated_image = requests.get(image_url).content  # download the image
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # Display the image in the default image viewer
        image = Image.open(image_path)
        image.show()
    
    # catch exceptions
    except openai.InvalidRequestError as err:
        print(err)

    ```

ဒီ code ကို ရှင်းပြပါမယ်-

- အရင်ဆုံး လိုအပ်တဲ့ library တွေကို import လုပ်ပါတယ်။ OpenAI library, dotenv library, requests library နဲ့ Pillow library တို့ပါဝင်ပါတယ်။

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    ```

- နောက်တစ်ခုက *.env* ဖိုင်ထဲက environment variable တွေကို load လုပ်ပါတယ်။

    ```python
    # import dotenv
    dotenv.load_dotenv()
    ```

- ထို့နောက် OpenAI API အတွက် endpoint, key, version နဲ့ type ကို သတ်မှတ်ပါတယ်။

    ```python
    # Get endpoint and key from environment variables
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY'] 

    # add version and type, Azure specific
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    ```

- နောက်တစ်ဆင့်မှာ ပုံကို generate လုပ်ပါတယ်-

    ```python
    # Create an image by using the image generation API
    generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
        size='1024x1024',
        n=2,
        temperature=0,
    )
    ```

    အထက်ပါ code က generate လုပ်ထားတဲ့ ပုံရဲ့ URL ပါဝင်တဲ့ JSON object တစ်ခုကို ပြန်ထုတ်ပေးပါတယ်။ ဒီ URL ကို သုံးပြီး ပုံကို download လုပ်နိုင်သလို ဖိုင်အဖြစ် သိမ်းထားနိုင်ပါတယ်။

- နောက်ဆုံးအနေနဲ့ ပုံကို ဖွင့်ပြီး ပုံကြည့်တဲ့ default viewer နဲ့ ပြသပါတယ်-

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ပုံဖန်တီးခြင်းနဲ့ ပတ်သက်တဲ့ အသေးစိတ်

```python
generation_response = openai.Image.create(

        n=2,
        temperature=0,

```

- **prompt** ဆိုတာက ပုံဖန်တီးဖို့ အသုံးပြုတဲ့ စာသား prompt ပါ။ ဒီဥပမာမှာ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ဆိုတဲ့ prompt ကို သုံးထားပါတယ်။
- **size** ဆိုတာက ဖန်တီးမယ့် ပုံရဲ့ အရွယ်အစားပါ။ ဒီမှာ 1024x1024 pixels အရွယ်အစားနဲ့ ပုံကို ဖန်တီးပါတယ်။
- **n** ဆိုတာက ဖန်တီးမယ့် ပုံအရေအတွက်ပါ။ ဒီဥပမာမှာ ပုံ ၂ ပုံ ဖန်တီးပါတယ်။
- **temperature** ဆိုတာက Generative AI model ရဲ့ output ကို မတည်မြဲမှု (randomness) ကို ထိန်းချုပ်တဲ့ parameter တစ်ခုပါ။ temperature က 0 နဲ့ 1 အကြားတန်ဖိုးတစ်ခုဖြစ်ပြီး 0 ဆိုရင် output က တိကျတည်မြဲပြီး 1 ဆိုရင် output က အလွတ်သဘောဖြစ်ပါတယ်။ မူလတန်ဖိုးက 0.7 ဖြစ်ပါတယ်။

ပုံနဲ့ ပတ်သက်ပြီး လုပ်နိုင်သေးတဲ့ အရာတွေကို နောက်အပိုင်းမှာ ဆက်လေ့လာပါမယ်။

## ပုံဖန်တီးခြင်းရဲ့ ထပ်မံနိုင်စွမ်းများ

Python နဲ့ အနည်းငယ်သော code များသုံးပြီး ပုံတစ်ပုံ ဖန်တီးနိုင်တာကို မြင်ခဲ့ပြီးပါပြီ။ ဒါပေမယ့် ပုံနဲ့ ပတ်သက်ပြီး လုပ်နိုင်သေးတဲ့ အရာတွေရှိပါတယ်။

သင်လုပ်နိုင်သေးတာတွေကတော့-

- **တည်းဖြတ်နိုင်ခြင်း**။ ရှိပြီးသား ပုံတစ်ပုံ၊ mask တစ်ခုနဲ့ prompt တစ်ခုကို ပေးပြီး ပုံကို ပြင်ဆင်နိုင်ပါတယ်။ ဥပမာအားဖြင့် ပုံတစ်ပုံရဲ့ တစ်စိတ်တစ်ပိုင်းမှာ အရာတစ်ခု ထပ်ထည့်နိုင်ပါတယ်။ ဥပမာ ကျွန်တော်တို့ရဲ့ ကြွက်ကလေးပုံမှာ ကြွက်ကလေးရဲ့ ခေါင်းပေါ်မှာ ဦးထုပ်တစ်လုံး ထပ်ထည့်နိုင်ပါတယ်။ ဒါလုပ်ဖို့ ပုံတစ်ပုံ၊ mask (ပြင်ချင်တဲ့နေရာကို ဖော်ပြတဲ့ mask) နဲ့ ဘာလုပ်ချင်တယ်ဆိုတဲ့ စာသား prompt တစ်ခု ပေးရပါမယ်။

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    မူလပုံမှာ ကြွက်ကလေးပဲ ပါဝင်မှာဖြစ်ပြီး နောက်ဆုံးပုံမှာတော့ ကြွက်ကလေးရဲ့ ခေါင်းပေါ်မှာ ဦးထုပ်ပါဝင်လာမှာပါ။
