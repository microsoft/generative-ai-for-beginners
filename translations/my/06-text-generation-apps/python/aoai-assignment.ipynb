# စာသားဖန်တီးခြင်းအက်ပ်များ ဖန်တီးပါ

ဒီ စာအုပ်တန်းထဲမှာ သင်တွေ့မြင်ခဲ့တာက prompt လို့ခေါ်တဲ့ အချက်အလက်တွေရှိပြီး "prompt engineering" လို့ခေါ်တဲ့ နယ်ပယ်တစ်ခုတည်းလည်း ရှိပါတယ်။ ChatGPT၊ Office 365၊ Microsoft Power Platform နဲ့ အခြားတစ်ချို့ ကိရိယာတွေက prompt တွေကို အသုံးပြု၍ တစ်စုံတစ်ခုကို ပြုလုပ်နိုင်အောင် ကူညီပေးပါတယ်။

အဲဒီလိုအတွေ့အကြုံကို အက်ပ်တစ်ခုမှာ ထည့်သွင်းချင်ရင် prompt, completion နဲ့ သင့်အလုပ်လုပ်ချင်တဲ့ library တစ်ခု ရွေးချယ်ဖို့ နားလည်ဖို့ လိုပါတယ်။ ဒီအခန်းမှာ မင်း တဲ့ အဲဒီအရာတွေလေ့လာပါမယ်။

## နိဒါန်း

ဒီအခန်းမှာ မင်းလုပ်မယ့်အရာတွေက:

- openai library နဲ့ ဒါရိုက်တာ အကြောင်းလေ့လာမယ်။
- openai ကိုသုံးပြီး စာသားဖန်တီးရေးအက်ပ်တစ်ခု တည်ဆောက်မယ်။
- prompt, temperature, tokens တွေလို concept တွေကို အသုံးပြုဖို့ နားလည်၊ စာသားဖန်တီးရေးအက်ပ် တည်ဆောက်မယ်။

## သင်ယူရမယ့် ရည်ရွယ်ချက်များ

ဒီသင်ခန်းစာနောက်ဆုံးမှာ မင်းအလုပ်လုပ်နိုင်မှာတွေက:

- စာသားဖန်တီးရေးအက်ပ်ဆိုတာဘာလဲ ဆိုတာ ရှင်းပြနိုင်မယ်။
- openai ကိုသုံးပြီး စာသားဖန်တီးရေးအက်ပ်တည်ဆောက်မယ်။
- အက်ပ်ကို tokens တိုး သို့မဟုတ် လျော့ပြီး temperature ကိုလည်းပြောင်းလဲအတွက် ပြင်ဆင်နိုင်မယ်၊ ထွက်ထားတာကလည်း မတူညီစေရမယ်။

## စာသားဖန်တီးရေးအက်ပ် ဆိုတာဘာလဲ။

သာမန်အက်ပ်တည်ဆောက်တဲ့အခါ အောက်ပါပုံစံ interface လေးတွေ ပါလာတတ်တယ်။

- Command-based။ Console app တွေမှာ command ရိုက်ပြီး လုပ်ဆောင်တယ်။ ဥပမာ `git` က command-based app တစ်ခု။
- User interface (UI)။ အက်ပ်တစ်ချို့မှာ graphical user interface (GUI) ရှိပြီး ခလုတ်နှိပ်၊ စာရိုက်၊ option ရွေးခွင့်တွေ ပါတယ်။

### Console နှင့် UI app များ၏ ကန့်သတ်ချက်များ

တစ်ခါ command-based app တစ်ခုရဲ့ command တွေ အတိုင်းထပ်ပြောရမယ်ဆိုရင်:

- **ကန့်သတ်ထားတယ်။** မင်းရိုက်ချင်တဲ့ command အားလုံးကိုမရိုက်နိုင်ဘူး၊ လက်ခံတဲ့ command တွေကိုသာ ရိုက်နိုင်တယ်။
- **ဘာသာစကားအတိအကျ**။ အက်ပ်တချို့က ဘာသာစကား များစွာကို ပံ့ပိုးပေးတတ်ပေမယ့် မူလကတော့ တစ်ခုချင်းစီကို အသုံးပြုဖို့ဖန်တီးထားတယ်။

### စာသားဖန်တီးရေးအက်ပ်များ၏ အကျိုးကျေးဇူးများ

စာသားဖန်တီးရေးအက်ပ်နဲ့ ဘယ်လို ကွာခြားသလဲ။

စာသားဖန်တီးရေးအက်ပ်မှာ commands တိတိတွဲမဟုတ်ဘူး၊ input language တိတိလည်း အကန့်အသတ်မရှိပါဘူး၊ သဘာဝဘာသာစကားနဲ့ ဆက်သွယ်နိုင်ပါတယ်။ ထို့အပြင် သင် interaction လုပ်နေလို့ရတဲ့ data source က အကြမ်းဖက်ကျယ်ပြန့်ဆုံး corpus တွေနဲ့ သင်ခန်းစာတွေပါရှိတာကြောင့်၊ ရိုးရိုး database က အကန့်အသတ်ရှိနေတဲ့အက်ပ်နဲ့ ကွာခြားပါတယ်။

### စာသားဖန်တီးရေးအက်ပ်နဲ့ ဘာတွေဆောက်လို့ရမလဲ။

အမျိုးမျိုးစိတ်ရှိပါတယ်။ ဥပမာ:

- **Chatbot တစ်ခု**။ သင့်ကုမ္ပဏီနဲ့ ထုတ်ကုန်တွေ အကြောင်း မေးခွန်းများကို ဖြေရှင်းပေးနိုင်တဲ့ chatbot ဖြစ်စေပါတယ်။
- **အကူအညီပေးသူ**။ LLM များက စာသားကို စုစည်းချုပ်နေရန်၊ အချက်အလက်တွေ ဆွဲထုတ်ဖော်ပြရန်၊ CV စာသားထုတ်ဖော်ရန် စသည့်အရာတွေလုပ်နိုင်ပါတယ်။
- **Code assistant**။ သင်သုံးတဲ့ language model ပေါ်မူတည်ပြီး ကုဒ်ရေးရမှာ ကူညီတဲ့ code assistant လုပ်နိုင်ပါတယ်။ ဥပမာ GitHub Copilot နဲ့ ChatGPT အသုံးပြုကာ ကုဒ်ရေးရာ အကူအညီရနိုင်ပါတယ်။

## ဘယ်လို စတင်နိုင်မလဲ။

LLM နဲ့ ပေါင်းစည်းဖို့ နည်းလမ်းနှစ်ခုရှိပြီး အောက်ပါအတိုင်း ဖြစ်တတ်ပါတယ်။

- API ကိုသုံးပါ။ ဒီမှာ prompt သုံးပြီး web request ဖန်တီးပြီးအလိုအလျောက်ထွက် စာသားကို ပြန်လည်ရရှိပါတယ်။
- Library ကိုသုံးပါ။ Libraries တွေက API calls တွေကို ခြုံငုံပြီး အသုံးပြုရ လွယ်ကူစေပါတယ်။

## Libraries/SDKs

LLM နှင့် အလုပ်လုပ်ရန် နာမည်ကြီး library တချို့ ရှိတယ်။

- **openai**၊ ဒီ library က မော်ဒယ်နှင့် ချိတ်ဆက်ရ အလွယ်ကူဆုံး ဖြစ်ပါတယ်။

အထက်ပိုင်းတွင် လုပ်ဆောင်တဲ့ library များလည်းရှိပါတယ်။

- **Langchain**။ Langchain က နာမည်ကြီးပြီး Python ကိုပံ့ပိုးပါတယ်။
- **Semantic Kernel**။ Semantic Kernel က Microsoft ရဲ့ library ဖြစ်ပြီး C#, Python, Java အတိုင်း ဘာသာစကားတွေကို ပံ့ပိုးပါတယ်။

## openai ကို အသုံးပြုပြီး ပထမဆုံး အက်ပ်

ပထမဆုံးအက်ပ်က ဘယ်လိုဖန်တီးရမလဲ၊ လိုအပ်သမျှ library တွေ ဘယ်လောက်လိုအပ်မလဲ စသဖြင့် ကြည့်ကြမယ်။

### openai ကို တပ်ဆင်ပါ

  > [!NOTE] ဒီအဆင့်က Codespaces သို့မဟုတ် Devcontainer ထဲမှာ notebook run တဲ့ အချိန် မလိုအပ်ပါ။


OpenAI နဲ့ Azure OpenAI ကို ဆက်သွယ်ဖို့ library များစွာရှိပါတယ်။ C#, Python, JavaScript, Java နဲ့ တခြား programming languages များလည်း အသုံးပြုလို့ရပါတယ်။  
ကျွန်တော်တို့က Python ရဲ့ `openai` library ကို ရွေးပြီး `pip` သုံးပြီး install လုပ်ပါမယ်။

```bash
pip install openai
```

Codespaces သို့မဟုတ် Dev Container မဟုတ်ဘဲ run တဲ့ ကိရိယာမှာ Python ကိုလည်း ထည့်သွင်းဖို့ လိုပါတယ်။

### Resource တစ်ခု ဖန်တီးပါ

မဖန်တီးထားရင် အောက်ပါ အဆင့်တွေ လုပ်ဆောင်ရပါမယ်။

- Azure မှာ အကောင့်တုန်းတစ်ခု ဖန်တီးပါ <https://azure.microsoft.com/free/>
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) မှာ Azure OpenAI resource တစ်ခု ဖန်တီးပါ။ [resource ဖန်တီးနည်း](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ကို ဒီလင့်က မြင်ပါ။


### API Key နဲ့ endpoint ကို ရှာပါ

အခုအချိန်မှာ `openai` library မှာ ဘယ် API key ကိုအသုံးပြုမလဲ ပြောပြဖို့လိုပါတယ်။ API key ရဖို့ Azure OpenAI resource ရဲ့ "Keys and Endpoint" ဘလိတ်ထဲသွားပြီး "Key 1" ကို ကူးယူပါ။

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ဒီအချက်အလက်တွေကို ကူးယူထားပြီးနောက် library တွေကို အသုံးပြုခွင့်ပြုပါမယ်။

> [!NOTE]
> API key ကို လိပ်စာအနေနဲ့ ကိုယ့်ကုဒ်ကနေ အလွဲသုံးမခံဖို့ကောင်းတာ ဖြစ်ပါတယ်။ Environment variables သုံးနည်းနဲ့သုံးပါ။
> - `AZURE_OPENAI_API_KEY` ကို environment variable အနေနဲ့ .env ဖိုင်ထဲသွင်းပါ။ ဒီသင်ခန်းစာရဲ့ မတိုင်ခင်ကျင့်သုံးထားပြီးဆိုရင် ပြီးစီးနေပါပြီ။


### Azure အတွက် configuration ပြုလုပ်ခြင်း

Azure OpenAI ကို အသုံးပြုတာဆိုရင် ဒီလို configuration ပြုလုပ်နိုင်ပါတယ်။ Responses API က Azure OpenAI **v1 endpoint** ကနေ လုပ်ဆောင်တာမို့ `OpenAI` client ကို `<your-endpoint>/openai/v1/` သို့ ဦးတည်ဆောင်ရွက်ပါမယ်။

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

အထက်မှာ အောက်ပါအတိုင်း သတ်မှတ်ထားတာပါ။

- `api_key` သည် Azure Portal မှာ ရှာတွေ့သည့် API key ဖြစ်သည်။
- `base_url` သည် Azure OpenAI endpoint ကို `/openai/v1/` ဖြင့် တွဲဖက်ထားသည်။ Azure Portal မှာ API key အနေဖြင့် Endpoint ကို ထပ်မံရှာဖွေနိုင်သည်။ v1 endpoint သုံးခြင်းဖြစ်သောကြောင့် `api_version` မလိုအပ်တော့ပါ။
- `deployment` သည် Foundry portal တွင် ဖန်တီးထားသော မော်ဒယ် deployment အမည်ဖြစ်သည်။

> [!NOTE]
> `os.environ` သည် environment variables များကို ဖတ်ရန် အသုံးပြုသော mapping ဖြစ်သည်။ `AZURE_OPENAI_API_KEY` နှင့် `AZURE_OPENAI_ENDPOINT` ကဲ့သို့ environment variables များ ဖတ်ရာတွင် သုံးနိုင်သည်။

## စာသား ဖန်တီးခြင်း

စာသားဖန်တီးဘို့ `responses.create` method ကို သုံးပါ။ ဥပမာက ဒီလိုပါ။

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

အထက်ရှိ ကုဒ်မှာ response object ကို ဖန်တီးပြီး သုံးမဲ့ မော်ဒယ်နဲ့ prompt ကို ထည့်သွင်းပါတယ်။ ပြီးတော့ ဖန်တီးထားသော စာသားကို print ထုတ်ထားပါသည်။

### Chat responses

Responses API ကို single-turn စာသားဖန်တီးခြင်းနဲ့ multi-turn chatbots နှစ်မျိုးလုံးအတွက် သင့်တော်ပြီး `input` list မှ ရှေ့ဆက် စကားဝိုင်းဖွဲ့လို ချိတ်ဆက် စာများ ပိုထည့်နိုင်တယ်။

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

ဒီစွမ်းဆောင်ရည်အကြောင်း ကိုက်ညီတဲ့ နောက်အခန်းမှာ လေ့လာမှာ ဖြစ်ပါတယ်။

## လေ့ကျင့်ခန်း - ဟင်းချက်နည်း များထုတ်လုပ်သူတစ်ခု တည်ဆောက်ခြင်း

အဏ်ကအနေဖြင့် ဒီအခြေအနေကို ဆော့ကစားပြီးပြီဆိုတော့၊ ပြသထားသော အခြေအနေနှင့် ကိုက်ညီသော ကုဒ်ကို ရေးစို့ကြစို့။ အဲဒီအတွက် အောက်ပါအဆင့်ဆင့်ကို လိုက်နာပါ။

1။ ရှိပြီးသား *app.py* ဖိုင်ကို အစပြုအနေနဲ့ အသုံးပြုပါ။
1။ `prompt` 변수ကို တွေ့ရှိပြီး အောက်ပါကုဒ်နဲ့ ပြောင်းလဲရေးပါ။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
