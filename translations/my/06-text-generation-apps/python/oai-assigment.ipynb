# စာသားဖန်တီးခြင်းအက်ပ်များ တည်ဆောက်ခြင်း

ဒီသင်ခန်းစာတွင် သင်မြင်တွေ့ခဲ့သလို prompt များနှင့် "prompt engineering" ဟုခေါ်သော စည်းမျဉ်းတစ်ခုရှိကြောင်း စတဲ့ အဓိက သဘောတရားများ ရှိသည်။ ChatGPT၊ Office 365၊ Microsoft Power Platform နဲ့ အခြား စွမ်းရည်မြင့်ကိရိယာများသည် ရည်ညွှန်းချက်များ အသုံးပြု၍ တစ်စုံတစ်ရာ ပြုလုပ်ရန် သင့်ကို ကူညီပေးကြသည်။

အက်ပ်တစ်ခုတွင် ဒီလို အတွေ့အကြုံကို ပေါင်းထည့်ရန် သင် prompt များ၊ completion များ အကြောင်းနားလည်ပြီး အသုံးပြုဖို့ စာကြည့်တိုက်တစ်ခု ရွေးချယ်ရမည်။ ဒီအခန်းမှာ သင် တဲ့အတိုင်း သင်ယူတော့မှာပါ။

## နိဒါန်း

ဒီအခန်းမှာ သင်လုပ်မယ့်အရာများမှာ:

- openai စာကြည့်တိုက်နှင့် ၎င်း၏ အဓိက သဘောတရားများကို သင်ယူမည်။
- openai ကို အသုံးပြု၍ စာသားဖန်တီးခြင်းအက်ပ် တစ်ခု တည်ဆောက်မည်။
- prompt၊ temperature နှင့် tokens စသည့် သဘောတရားများဖြင့် စာသားဖန်တီးခြင်းအက်ပ် တည်ဆောက်ရန် နားလည်မည်။

## စ Learning လို့ရအောင် သင်ယူရမည့်အချက်များ

ဒီသင်ခန်းစာအဆုံးသတ်တဲ့အချိန်မှာ သင်လုပ်နိုင်မှာတွေက:

- စာသားဖန်တီးခြင်းအက်ပ်ဆိုတာဘာလဲ ဆိုတာ ရှင်းပြနိုင်မယ်။
- openai ကိုအသုံးပြု၍ စာသားဖန်တီးခြင်းအက်ပ် တည်ဆောက်နိုင်မယ်။
- သင့်အက်ပ်ကို tokens ပိုမို့ ဒါမှမဟုတ် လျော့နည်းစေနိုင်ရန် နှင့် temperature ကို ပြောင်းလဲနိုင်ပြီး မတူညီသော အထွက်ရှိစေရန် ပြင်ဆင်နိုင်မယ်။

## စာသားဖန်တီးခြင်းအက်ပ်ဆိုတာ ဘာလဲ?

ပုံမှန်အားဖြင့် အက်ပ်တစ်ခုတည်ဆောက်တယ်ဆိုရင် အောက်ပါကဲ့သို့ မျက်နှာပြင်တစ်ခုရှိပါတယ်။

- Command-based။ Console app များသည် စာမေးပုံဖြင့် command တစ်ခုကို ရိုက်ထည့်ရုံဖြင့် လုပ်ဆောင်နိုင်တဲ့ အက်ပ်မျိုးဖြစ်ပြီး ဥပမာအားဖြင့် `git` သည် command-based app တစ်ခုဖြစ်သည်။
- User interface (UI)။ အက်ပ်တချို့မှာ graphical user interface (GUI) ရှိပြီး ခလုတ်တွင်နှိပ်ရုံ၊ စာသားရိုက်ထည့်ရန်၊ ရွေးချယ်မှု ရှိသည်။

### Console နှင့် UI အက်ပ်များမှာ ကန့်သတ်ချက်ရှိသည်

Command-based အက်ပ်တစ်ခုတွေနဲ့ နှိုင်းယှဉ်ပါ၊ command ကို ရိုက်ထည့်တယ်ဆိုတော့:

- **ကန့်သတ်ထားသည်**။ သင် ဘာ command မဆို ရိုက်ထည့်၍ မရပါ၊ အက်ပ်မှထောက်ခံသော command များသာ ရိုက်ထည့်နိုင်သည်။
- **ဘာသာစကား အထူးပြု**။ အက်ပ်တချို့မှာ ဘာသာစကားများစွာ ထောက်ခံပေမဲ့ ပုံမှန်အားဖြင့် အက်ပ်သည် တစ်ခုတည်းသော ဘာသာစကားအတွက် ဖန်တီးထားပြီး နောက်ထပ်ဘာသာစကားထောက်ခံမှု ပေါင်းထည့်နိုင်သည်။

### စာသားဖန်တီးခြင်းအက်ပ်များ၏ အကျိုးကျေးဇူးများ

စာသားဖန်တီးခြင်းအက်ပ်က ဘာများကွဲပြားသလဲ?

စာသားဖန်တီးခြင်းအက်ပ်မှာ သင်မှာ ပိုမိုတင်းကြပ်မှုနည်းပြီး Command များ သို့မဟုတ် ဘာသာစကားအစား သဘာဝဘာသာစကားဖြင့် အက်ပ်နှင့် ဆက်သွယ်နိုင်သည်။ နောက်ထပ် အကျိုးအမြတ်ကတော့ သင်ဖြစ်ပြီးသား သင်ထိတွေ့နေတဲ့ ဒေတာရင်းမြစ်မှာ ကျယ်ပြန့်သော အချက်အလက်စုစည်းထားပြီး ပုံမှန်အက်ပ်ဟာ database မှာသာ အကန့်အသတ်ရှိတတ်တာပါ။

### စာသားဖန်တီးခြင်းအက်ပ်နှင့် ဘာတွေ ဖန်တီးနိုင်မလဲ?

ဖန်တီးနိုင်သည့် အရာများစွာရှိသည်။ ဥပမာအားဖြင့်:

- **Chatbot**။ သင့်ကုမ္ပဏီနှင့် ထုတ်ကုန်များအကြောင်းမေးခွန်းများကို ဖြေကြားတတ်သော chatbot ကောင်းတစ်ခု ဖြစ်နိုင်သည်။
- **သူငယ်ချင်းအကူအညီ**။ LLM များက စာသားကို ရှုခင်း၊ အထွေထွေအချက်အလက်ရယူခြင်း၊ ရုပ်သံကြောင်းဖြင့် စာသားဖန်တီးခြင်း အစရှိသည့် အလုပ်များတွင် အထူးကောင်းမွန်သည်။
- **ကုဒ်အကူအညီ**။ သင် အသုံးပြုသော ဘာသာစကားမော်ဒယ်ပေါ်မူတည်၍ ကုဒ်ရေးရန် ကူညီပေးသော ကူညီရေး အက်ပ် တစ်ခု ဖန်တီးနိုင်သည်။ ဥပမာအားဖြင့် GitHub Copilot နဲ့ ChatGPT ကို အသုံးပြု၍ ကုဒ်ရေးရာတွင် ကူညီနိုင်သည်။

## ဘယ်လို စတင်ရမလဲ?

LLM အတွက် ပေါင်းသင်းလုပ်ဆောင်ရန် နည်းလမ်း နှစ်မျိုးရှိပြီး အောက်ပါ နည်းလမ်းများ ဖြစ်သည်။

- API ကို အသုံးပြုခြင်း။ ဒီမှာ prompt နှင့်အတူ web requests ဖန်တီးပြီး စာသား ဖန်တီးမှုကို ပြန်ရမှာဖြစ်သည်။
- စာကြည့်တိုက်အသုံးပြုခြင်း။ စာကြည့်တိုက်များသည် API ခေါ်ဆိုမှုများ ကို ရှယ်ကောက်ပြီး အသုံးပြုရလွယ်ကူစေသည်။

## စာကြည့်တိုက်များ/SDK များ

LLM များနှင့်လက်တွဲဆက်သွယ်ရာတွင် ကျော်ကြားသော စာကြည့်တိုက် အနည်းငယ်ရှိသည်။

- **openai**၊ ဒီစာကြည့်တိုက်သည် မော်ဒယ်နှင့် ချိတ်ဆက်၍ prompt များ ကို လွယ်ကူစွာ ပို့လို့ရသည်။

ထို့နောက် အဆင့်မြင့် စာကြည့်တိုက်များ ချရေးဖန်တီးထားသည်မှာ:

- **Langchain**။ Langchain သည် ကျော်ကြားပြီး Python ကို ထောက်ခံသည်။
- **Semantic Kernel**။ Semantic Kernel သည် Microsoft မှ ထုတ်ထားသော စာကြည့်တိုက်ဖြစ်ပြီး C#၊ Python နဲ့ Java ဘာသာစကားများကို စောင့်ပို့သည်။

## openai ကို အသုံးပြု၍ ပထမ အက်ပ်

ပထမဆုံးအက်ပ်ကို ဘယ်လိုတည်ဆောက်ရမလဲ၊ ဘာစာကြည့်တိုက်လိုအပ်ပြီး မည်မျှလိုမလဲ စသည်တို့ကို ကြည့်ရအောင်။

### openai ကို ထည့်သွင်းခြင်း

  > [!NOTE] Codespaces သို့မဟုတ် Devcontainer တွင် ဒီ notebook ကို ပြေးမရင် ဒီအဆင့် လိုအပ်အံ႔မယ်။


OpenAI သို့မဟုတ် Azure OpenAI နှင့် ဆက်သွယ်ရန် စာကြည့်တိုက်များ များပြားပြီး C#၊ Python၊ JavaScript၊ Java စတဲ့ ဘာသာစကားများကို အသုံးပြုနိုင်ပါသည်။
Openai Python စာကြည့်တိုက်ကိုရွေးချယ်ပြီး pip ဖြင့် ထည့်သွင်းပါမယ်။

```bash
pip install openai
```

Codespaces သို့မဟုတ် Dev Container တွင် မပြေးပါက သင့်ကွန်ပျူတာတွင် [Python](https://www.python.org/) ထည့်သွင်းထားရန် လိုအပ်သည်။

### အရင်းအမြစ် တစ်ခု ဖန်တီးပြီး API key ကို ရှာဖွေရန်

မရှိသေးကွောင့် အောက်ပါလုပ်ငန်းစဉ်များ ဆောင်ရွက်ရမည်။

- OpenAI <https://platform.openai.com/signup> တွင် အကောင့်ဖွင့်ပါ။
- ယခု API keys များ ရယူပါ <https://platform.openai.com/api-keys>။

>[!NOTE]
> သင့် API key ကို ကုဒ်မှ ကွဲထွက်ရောင်းရန် ဦးစားပေးသင့်သည်။ environment variables ဖြင့် ပြုလုပ်နိုင်သည်။
> - သင့် .env ဖိုင်ထဲတွင် environment variable `OPENAI_KEY` ကို သင့် API key ဖြင့် သတ်မှတ်ပါ။ ဒီသင်ခန်းစာ၏ မူလ လေ့ကျင့်မှုများပြီးဆုံးထားလျှင် ပြီးပြည့်စုံပါတယ်။
> - API Key သည် တစ်ခါတည်းသာ ရရှိနိုင်သောကြောင့် မှန်ကန်စွာ ကူးယူထားကြောင်း စစ်ဆေးထားသင့်သည်။ လိုအပ်လျှင် key ကို ဖျက်ပြီး အသစ်ဖန်တီးသင့်သည်။


### OpenAI အတွက် ပြင်ဆင်မှု ဆက်တင်

OpenAI ကို အသုံးပြုလျှင် ပြင်ဆင်ခြင်း ဒီလိုလုပ်ပါ။

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

အထက်မှာ ကျွန်တော်တို့ သတ်မှတ်ထားတာသည်

- `api_key`၊ သင့် OpenAI dashboard မှ ရနိုင်သော API key ဖြစ်သည်။
- `deployment`၊ သင့် GPT ဗားရှင်းဖြစ်သည်။

> [!NOTE]
> `os.environ` သည် environment variables ကို ဖတ်ရှုနိုင်သော function ဖြစ်သည်။ `OPENAI_API_KEY` ကဲ့သို့သော environment variable များ ဖတ်ရန် အသုံးပြုနိုင်သည်။

## စာသား ဖန်တီးခြင်း

စာသားဖန်တီးရန် `responses.create` နည်းလမ်းကို အသုံးပြုသည်။ နမူနာက ဒီလိုဖြစ်သည် -

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

အထက်ပါ ကိုဒ်တွင် response object တစ်ခု ဖန်တီး၍ သုံးမယ့် မော်ဒယ်နဲ့ prompt ကို ဖြည့်သွင်းပါတယ်။ ထို့နောက် ဖန်တီးပြီးစာသားကို ပုံနှိပ်ပေးသည်။

### Chat ဖြေကြားချက်များ

ယခုထိ တစ်ခြား prompt တစ်ခုမှ စာသားဖန်တီးခြင်းကို မြင်ရပြီ။ Responses API သည် chatbot အတွက်လည်း သင့်တော်သည် - `input` အဖြစ် မက်ဆေ့ချ်များ စုစည်းပေး၍ စကားပြောဆိုမှု တည်ဆောက်နိုင်သည်။ အသုံးပြုနည်းနမူနာပါ။

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

ဒီလုပ်ဆောင်ချက်အကြောင်း ပိုမိုအသေးစိတ် နောက်ဆုံး အခန်းတစ်ခုတွင် လေ့လာမည်။

## အမျိုးမျိုးသောအကြောင်းအရာများအတွက် အမျိုးမျိုးသော prompt မျိုးများ

Prompt များကို အမျိုးမျိုးသောတာဝန်များအတွက် အသုံးပြုနိုင်သည်။ ဥပမာအားဖြင့် -

- **စာသားအမျိုးအစားတစ်မျိုး ထုတ်ပေးခြင်း**။ ဥပမာ အတွက် ကဗျာ၊ ကွစ်ရှေးမေးခွန်းများ စသည်တို့ ထုတ်ပေးနိုင်သည်။
- **သတင်းအချက်အလက်ရှာဖွေခြင်း**။ ဥပမာ 'ဝက်ဘ်ဖွံ့ဖြိုးတိုးတက်မှုမှာ CORS ဆိုတာဘာလဲ?' ဆိုတာကို ရှာဖွေဖို့ Prompt အသုံးပြုနိုင်သည်။
- **ကုဒ် (code) ထုတ်ပေးခြင်း**။ မေးခွန်းတစ်ခုအနေဖြင့် သင်သည် email ကို အတည်ပြုရန် regular expression တစ်ခု ဖန်တီးရန် သို့မဟုတ် ဝက်ဘ်အက်ပလီကေးရှင်းတစ်ခု စသည်တို့ကို ဖန်တီးရန် Prompt ကို အသုံးပြုနိုင်သည်။

## လေ့ကျင့်မှု - အစားအစာမွန်းဆအက်ပလီကေးရှင်း တီထွင်ပါ

အခု စာရိုက်ပြထားတဲ့ အခြေအနေကို ကျွန်တော်တို့ပြသပြပြီးသား ဖြစ်တဲ့အတွက်၊ ပြသထားတဲ့ အခြေအနေကို ကိုက်ညီအောင် ကုဒ်ရေးကြပါစို့။ အဲဒီအတွက် လိုက်နာရမယ့် အဆင့်တွေကတော့ -

1. ရှိပြီးသား *app.py* ဖိုင်ကို စတင်အနေနဲ့ သုံးပါ
1. `prompt` ဟုခေါ်သော variable ကို ရှာပြီး အတွက်ကုဒ်ကို အောက်ပါအတိုင်း ပြောင်းလဲရေးပါ -


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> မှတ်ချက်၊ သင့် LLM သည် အတိအကျ မရှိသောကြောင့် ပြေးပြေးတိုင်း မတူညီသော ရလဒ်များ ရရှိနိုင်ပါသည်။

1. အောက်ပါအတိုင်း ကုဒ်ကိုပြောင်းလဲကြမယ်။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **အစိတ်အပိုင်းများကို စစ်ထုတ်ခြင်း** ။ ကြိုက်နှစ်သက်မှုမရှိသည့် သို့မဟုတ် allergy ရှိသော အစားအစာများကို စစ်ထုတ်နိုင်ရန်လိုအပ်သည်။ ဤပြောင်းလဲမှုပြုရန်အတွက်၊ ကျွန်ုပ်တို့ အပိုမိုလက်ခံရန် prompt ကို ပြင်ဆင်ပြီး အဆုံးတွင် filter condition တစ်ခု ထည့်သွင်းနိုင်ပါသည်။

    မြင်နိုင်သည့်အတိုင်း၊ ပျားနို့ ပါဝင်သည့် recipe များအားလုံး လုပ်ေဆာင့ွရန် ဖယ်ရှားထားပါသည်။ သို့သော် lactose intolerant ဖြစ်ပါက၊ ချိစ် ပါဝင်သည့် recipe များကိုလည်း ဖယ်ရှားလိုပါက အကြောင်းအရာကို ပိုမိုရှင်းလင်းစေရန် လိုအပ်ပါသည်။

- **Shopping list ထုတ်ပေးခြင်း** ။ အိမ်တွင်းရှိသော ပစ္စည်းများကိုလည်း ထည့်သွင်းစဉ်းစား၍ shopping list ထုတ်ပေးလိုပါသည်။

    ဤလုပ်ဆောင်ချက်အတွက် တစ်ချက်တည်း prompt ဖြင့် ပြီးစီးနိုင်သလို ဒါမှမဟုတ် နှစ်ချက် ခြား၍ prompt နှစ်ခုပြုလုပ်နိုင်ပါသည်။ နောက်ဆုံးနည်းလမ်းကို စမ်းပါမည်။ ဤနေရာတွင် ထပ်မံ prompt တစ်ခု ထည့်သွင်းရန် အကြံပြုထားပြီး ၎င်းအတွက် အရင် prompt ၏ရလဒ်ကို နောက် prompt ၏ context အဖြစ် ထည့်သွင်းရန် လိုအပ်ပါသည်။

    ပထမ prompt ၏ ရလဒ်ကို print ထုတ်နေသည့် ကုဒ်၏ အပိုင်းကိုရှာဖွေပြီး အောက်ပါကုဒ်ကို ထည့်သွင်းပါ။
    
    ```python
    old_prompt_result = response.output_text

    new_prompt = f"{old_prompt_result} {prompt}"

    ```

    အောက်ပါအချက်များကို သတိပြုပါ။

    - ပထမ prompt ၏ရလဒ်ကို သစ် prompt ထဲ ပေါင်းထည့်၍ prompt အသစ်ကို တည်ဆောက်နေပါသည်။


- **token အရှည်အကြောင်းအရာ** ။ စိတ်တိုင်းကျ ကျွန်ုပ်တို့ မျှော်လင့်သော စာသား ထုတ်ဝေနိုင်ရန် token အရေအတွက်ကို စဉ်းစားသင့်ပါသည်။ token များသည် ငွေကြေးကုန်ကျစရိတ်ရှိသောကြောင့် ဖြစ်နိုင်သမျှ token အရေအတွက်ကို စိုက်စားသုံးစွဲရန် ကြိုးစားသင့်သည်။ ဥပမာ prompt ကို ဂရုစိုက်စွာ ဖန်းတီး၍ token အနည်းငယ်သုံးစွဲမည့် ရလဒ်ရနိုင်မလား စဉ်းစားပါ။

   token များအသုံးပြုမှုကိုပြောင်းလဲရန် `max_output_tokens` parameter ကို အသုံးပြုနိုင်သည်။ ဥပမာ token ၁၀၀ သုံးရန် အလိုရှိပါက အောက်ပါအတိုင်း လုပ်နိုင်ပါသည်။

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **temperature ဖြင့် သရုပ်ပြခြင်း** ။ Temperature သည် ယခင်တွင်မဖော်ပြခဲ့သော်လည်း ကျွန်ုပ်တို့ အစီအစဉ် ဖော့ရှင်ဘယ်လိုရှိမည်ကို သက်ဆိုင်ရာ စာချုပ် တစ်ခုဖြစ်သည်။ temperature တန်ဖိုးနိမ့်သည် output ကို ပိုမို ကြိုတင်ခန့်မှန်းနိုင်ပြီး temperature တန်ဖိုး မြင့်သည် output ကို ပိုမို အလွယ်တကူအသစ်ပြောင်းလဲမှုရှိစေသည်။ ထွက်ရှိမည့် အမျိုးအစားအမျိုးမျိုးရှိသင့်မရှိ သုံးသပ်ပါ။
