# စာသားထုတ်လုပ်ရေးအက်ပ်များ တည်ဆောက်ခြင်း

ဒီသင်ရိုးတစ်လျှောက်မှာ သင်မြင်တွေ့ခဲ့သလို prompt တွေလို အဓိကအယူအဆတွေနဲ့ "prompt engineering" ဆိုတဲ့ အထူးအတန်းတစ်ခုတောင်ရှိပါတယ်။ ChatGPT, Office 365, Microsoft Power Platform တို့လို သုံးနိုင်တဲ့ကိရိယာတွေမှာလည်း prompt တွေသုံးပြီး တစ်ခုခုလုပ်ဆောင်နိုင်အောင် ထောက်ပံ့ပေးထားပါတယ်။

ဒီလိုအတွေ့အကြုံကို သင့်အက်ပ်ထဲထည့်ချင်ရင် prompt, completion တို့လို အယူအဆတွေနဲ့ သုံးမယ့် library ကိုရွေးချယ်နားလည်ဖို့လိုပါတယ်။ ဒီအခန်းမှာ အဲဒီအကြောင်းတွေကို သင်လေ့လာရပါမယ်။

## မိတ်ဆက်

ဒီအခန်းမှာ သင်:

- openai library နဲ့ အဓိကအယူအဆတွေကို လေ့လာရမယ်။
- openai ကိုသုံးပြီး စာသားထုတ်လုပ်တဲ့အက်ပ်တစ်ခု တည်ဆောက်ရမယ်။
- prompt, temperature, tokens တို့လို အယူအဆတွေကို သုံးပြီး စာသားထုတ်လုပ်ရေးအက်ပ်တစ်ခု ဘယ်လိုတည်ဆောက်ရမလဲ နားလည်လာမယ်။

## သင်ယူရမယ့်ရည်မှန်းချက်များ

ဒီသင်ခန်းစာအဆုံးမှာ သင်:

- စာသားထုတ်လုပ်ရေးအက်ပ်ဆိုတာ ဘာလဲ ဆိုတာရှင်းပြနိုင်မယ်။
- openai ကိုသုံးပြီး စာသားထုတ်လုပ်ရေးအက်ပ်တစ်ခု တည်ဆောက်နိုင်မယ်။
- သင့်အက်ပ်မှာ tokens များ/နည်း သုံးနိုင်အောင်၊ temperature ကိုလည်း ပြောင်းလဲနိုင်အောင် ပြင်ဆင်နိုင်မယ်။

## စာသားထုတ်လုပ်ရေးအက်ပ်ဆိုတာဘာလဲ?

ပုံမှန်အက်ပ်တစ်ခု တည်ဆောက်တဲ့အခါ အောက်ပါအတိုင်း interface တစ်ခုခုရှိပါတယ် -

- အမိန့်ပေးသည့်ပုံစံ (Command-based)။ Console app တွေက သင် command တစ်ခုရိုက်ထည့်ပြီး တစ်ခုခုလုပ်ဆောင်တဲ့ app တွေပါ။ ဥပမာ `git` က command-based app တစ်ခုပါ။
- အသုံးပြုသူအင်တာဖေ့စ် (UI)။ GUI ပါတဲ့ app တွေမှာ button တွေကိုနှိပ်တာ၊ စာရိုက်ထည့်တာ၊ ရွေးချယ်မှုတွေပြုလုပ်တာတွေပါဝင်ပါတယ်။

### Console နဲ့ UI app တွေက အကန့်အသတ်ရှိတယ်

Command-based app တစ်ခုနဲ့ နှိုင်းယှဉ်ကြည့်ရအောင် -

- **အကန့်အသတ်ရှိတယ်**။ သင် app ကထောက်ပံ့တဲ့ command တွေကိုသာ ရိုက်ထည့်နိုင်ပြီး၊ မည်သည့် command မဆို ရိုက်လို့မရပါ။
- **ဘာသာစကားအထူးပြု**။ App တချို့က ဘာသာစကားအမျိုးမျိုးထောက်ပံ့ပေမယ့် မူလတည်ဆောက်တဲ့အခါမှာ တစ်ခုတည်းသောဘာသာစကားအတွက်သာ ဖြစ်တတ်ပါတယ်။

### စာသားထုတ်လုပ်ရေးအက်ပ်တွေရဲ့ အားသာချက်များ

စာသားထုတ်လုပ်ရေးအက်ပ်က ဘာကြောင့်ကွဲပြားသလဲ?

စာသားထုတ်လုပ်ရေးအက်ပ်မှာ သင်ပိုပြီးလွတ်လပ်မှုရှိပါတယ်။ Command တစ်စုတစ်စည်းနဲ့ သို့မဟုတ် input ဘာသာစကားတစ်ခုနဲ့ ကန့်သတ်ထားတာမဟုတ်ပါဘူး။ သင့်အနေနဲ့ သဘာဝဘာသာစကားနဲ့ app ကို ဆက်သွယ်နိုင်ပါတယ်။ နောက်ထပ်အားသာချက်ကတော့ သင်သုံးနေတဲ့ data source က အကြီးမားဆုံးသော သတင်းအချက်အလက်များနဲ့ already သင်ကြားထားတဲ့အတွက်၊ ရိုးရာ app တစ်ခုလို database ထဲမှာရှိတာနဲ့ ကန့်သတ်မထားဘူးဆိုတာပါ။

### စာသားထုတ်လုပ်ရေးအက်ပ်နဲ့ ဘာတွေတည်ဆောက်နိုင်လဲ?

တည်ဆောက်နိုင်တာတွေ များစွာရှိပါတယ်။ ဥပမာ -

- **Chatbot**။ သင့်ကုမ္ပဏီနဲ့ ထုတ်ကုန်တွေကို မေးခွန်းဖြေတဲ့ chatbot တစ်ခုတည်ဆောက်နိုင်ပါတယ်။
- **Helper**။ LLM တွေက စာသားအကျဉ်းချုပ်ရေးခြင်း၊ အချက်အလက်ထုတ်ယူခြင်း၊ CV တို့လို စာသားထုတ်လုပ်ခြင်းမှာ အထူးကောင်းပါတယ်။
- **Code assistant**။ သုံးမယ့် language model ပေါ်မူတည်ပြီး code assistant တစ်ခုတည်ဆောက်နိုင်ပါတယ်။ ဥပမာ GitHub Copilot, ChatGPT တို့ကို code ရေးရာမှာ အကူအညီရနိုင်ပါတယ်။

## ဘယ်လိုစတင်ရမလဲ?

LLM နဲ့ ပေါင်းစည်းဖို့ နည်းလမ်းနှစ်ခုရှိပါတယ် -

- API သုံးခြင်း။ ဒီမှာ သင့် prompt နဲ့ web request တစ်ခုတည်ဆောက်ပြီး ပြန်လာတဲ့ generated text ကိုရပါတယ်။
- Library သုံးခြင်း။ Library တွေက API ခေါ်တာတွေကို လွယ်ကူအောင် encapsulate လုပ်ပေးပါတယ်။

## Libraries/SDKs

LLM တွေနဲ့ အလုပ်လုပ်ဖို့ နာမည်ကြီး library တွေရှိပါတယ် -

- **openai** - ဒီ library က သင့် model နဲ့ ချိတ်ဆက်ပြီး prompt တွေ ပို့ဖို့ လွယ်ကူစေပါတယ်။

ထို့အပြင် အဆင့်မြင့် library တွေလည်းရှိပါတယ် -

- **Langchain** - Langchain က နာမည်ကြီးပြီး Python ကိုထောက်ပံ့ပါတယ်။
- **Semantic Kernel** - Microsoft ရဲ့ library ဖြစ်ပြီး C#, Python, Java တို့ကိုထောက်ပံ့ပါတယ်။

### Resource တစ်ခုဖန်တီးပြီး API key ကို ရှာပါ

မလုပ်ရသေးဘူးဆိုရင် အောက်ပါအဆင့်တွေ လုပ်ပါ -

- OpenAI မှာ အကောင့်ဖွင့်ပါ <https://platform.openai.com/signup>။

>[!NOTE]
> သင့် API key ကို သင့် code နဲ့ ခွဲထားသင့်ပါတယ်။ အဲဒါအတွက် environment variable သုံးနိုင်ပါတယ်။
> - `OPENAI_KEY` ဆိုတဲ့ environment variable ကို သင့် .env ဖိုင်ထဲမှာ သင့် API key နဲ့ သတ်မှတ်ပါ။ ဒီသင်တန်းရဲ့ ယခင်လေ့ကျင့်ခန်းတွေ ပြီးဆုံးထားရင် အားလုံးအဆင်သင့်ဖြစ်နေပါပြီ။
> - API Key ကို တစ်ကြိမ်တည်းသာ ရနိုင်တာမို့ မှန်ကန်စွာ copy လုပ်ထားတာ သေချာစစ်ဆေးပါ။ မအောင်မြင်ဘူးဆိုရင် key ကို ဖျက်ပြီး အသစ်ထုတ်ယူပါ။

### OpenAI ကို ပြင်ဆင်သတ်မှတ်ခြင်း

OpenAI ကိုသုံးမယ်ဆိုရင် အောက်ပါအတိုင်း ပြင်ဆင်နိုင်ပါတယ် -

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

အထက်မှာ သတ်မှတ်ထားတာတွေက -

- `api_key` - OpenAI dashboard မှာ ရနိုင်တဲ့ သင့် API key ပါ။
- `deployment` - သင်သုံးမယ့် GPT version ပါ။

> [!NOTE]
> `os.environ` က environment variable တွေကို ဖတ်တဲ့ function ပါ။ `OPENAI_API_KEY` လို environment variable တွေကို ဖတ်ဖို့ သုံးနိုင်ပါတယ်။

## စာသားထုတ်လုပ်ခြင်း

စာသားထုတ်လုပ်ဖို့ `chat.completion` class ကိုသုံးပါတယ်။ ဥပမာ -

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

အထက်ပါ code မှာ completion object တစ်ခုဖန်တီးပြီး သုံးမယ့် model နဲ့ prompt ကို ပေးပို့ပါတယ်။ ပြီးရင် ထုတ်လုပ်လာတဲ့ စာသားကို print ထုတ်ပါတယ်။

### Chat completions

ယခင်က `Completion` ကိုသုံးပြီး စာသားထုတ်လုပ်တာတွေ့ခဲ့ပါပြီ။ ဒါ့အပြင် chatbot တွေအတွက် ပိုသင့်တော်တဲ့ `ChatCompletion` ဆိုတဲ့ class တစ်ခုလည်းရှိပါတယ်။ ဥပမာ -

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

ဒီအကြောင်းကို နောက်အခန်းမှာ နောက်ထပ်လေ့လာရပါမယ်။

## လေ့ကျင့်ခန်း - သင့်ပထမဆုံး စာသားထုတ်လုပ်ရေးအက်ပ်

OpenAI service ကို ဘယ်လိုပြင်ဆင်သုံးစွဲရမလဲ လေ့လာပြီးပြီဆိုတော့ အခု သင့်ပထမဆုံး စာသားထုတ်လုပ်ရေးအက်ပ်ကို တည်ဆောက်ချိန်ရောက်ပါပြီ။ အက်ပ်တည်ဆောက်ဖို့ အောက်ပါအဆင့်တွေလိုက်နာပါ -


1. virtual environment တစ်ခု ဖန်တီးပြီး openai ကို install လုပ်ပါ။

  > [!NOTE] ဒီအဆင့်ကို Codespaces မှာ သို့မဟုတ် Devcontainer အတွင်းမှာ ဒီ notebook ကို run လုပ်မယ်ဆိုရင် မလိုအပ်ပါ။


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> သင် Windows ကို အသုံးပြုနေပါက `venv\Scripts\activate` ကို ရိုက်ထည့်ပါ၊ `source venv/bin/activate` အစား အသုံးပြုပါ။

> [!NOTE]
> သင့် OpenAI key ကို ရှာဖွေရန် https://platform.openai.com/settings/organization/api-keys သို့ သွားပြီး `API keys` ကို ရှာပါ။ အဲဒီမှာ key အသစ်တစ်ခု ဖန်တီးနိုင်ပြီး တန်ဖိုးကို ချက်ချင်း ကူးယူနိုင်ပါသည်။


1. *app.py* ဖိုင်တစ်ခု ဖန်တီးပြီး အောက်ပါကုဒ်ကို ထည့်ပါ။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
    
    # print response
    print("Shopping list:")
    print(completion.choices[0].message.content)
    ```
