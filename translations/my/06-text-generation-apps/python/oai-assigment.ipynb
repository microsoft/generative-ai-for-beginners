# စာသား ဖန်တီးခြင်း အက်ပ်များ တည်ဆောက်ခြင်း

ဒီ သင်ရိုးကြောင်းအတန်းမှာ အထောက်အထားများဖြစ်တဲ့ prompt များနဲ့ "prompt engineering" ဆိုတဲ့ သင်ခန်းစာတစ်ခုလုံးကဲ့သို့သော အခြေခံအယူတွေကို သင်မြင်ရပြီဖြစ်ပါတယ်။ ChatGPT, Office 365, Microsoft Power Platform တို့ကဲ့သို့ သင်နှင့်အတူ လုပ်ဆောင်နိုင်သော ကိရိယာများစွာက prompt များကို အသုံးပြုရန် ထောက်ပံ့ပေးသည်။

အက်ပ်တစ်ခုမှာ ဒီလို အတွေ့အကြုံတစ်ရပ် ထည့်သွင်းချင်ရင်တော့ prompt, completion တို့ကဲ့သို့ အယူအဆများကို နားလည်ရန်လိုပြီး အသုံးပြုမည့် ဇယားတစ်ခုကို ရွေးချယ်ဖို့ လိုအပ်ပါတယ်။ ဒါဟာ ဤအခန်းမှာ သင်ယူသွားမဲ့အကြောင်းပါ။

## မိတ်ဆက်

ဒီအခန်းမှာ သင်လုပ်ဆောင်မည့်အရာများမှာ-

- openai စာကြည့်တိုက်နဲ့ ၎င်း၏ အဓိက အယူအဆများကို လေ့လာခြင်း။
- openai ကို အသုံးပြု၍ စာသား ဖန်တီးခြင်း အက်ပ်တစ်ခု တည်ဆောက်ခြင်း။
- prompt, temperature, token များကဲ့သို့ အယူအဆများကို လေ့လာ၍ စာသားဖန်တီး အက်ပ်တည်ဆောက်ခြင်း။

## သင်ယူရမည့် ရည်မှန်းချက်များ

ဒီသင်ခန်းစာ အဆုံးတွင် သင်အနိုင်ရမယ့်အရာများမှာ-

- စာသား ဖန်တီး အက်ပ် ဆိုတာဘာလဲ ဆိုတာ ကို ရွေးပြောနိုင်သည်။
- openai ကို အသုံးပြု၍ စာသား ဖန်တီးအက်ပ်တစ်ခု တည်ဆောက်နိုင်သည်။
- ပိုမိုသို့မဟုတ် နည်းနည်းသာ token များကို အသုံးပြုရန် နှင့် temperature ကို ပြောင်းလဲ၍ ပါဝင်မှုမျိုးစုံကို ရရှိစေရန် သတ်မှတ်နိုင်သည်။

## စာသား ဖန်တီး အက်ပ် ဆိုတာဘာလဲ?

ဘာသာပြန်အားဖြင့် အက်ပ်တစ်ခု ဖန်တီးတဲ့အခါ ပုံမှန်အားဖြင့် အောက်ပါအတိုင်း UI တစ်ခုရှိတတ်သည်-

- command-based: command ရိုက်ထည့်ပြီး တာဝန်တစ်ခုဖြစ်အောင် ဆောင်ရွက်တဲ့ console app များ ဖြစ်တတ်သည်။ ဥပမာအားဖြင့် `git` သည် command-based app ဖြစ်သည်။
- User interface (UI): အချို့ app များမှာ ခလုတ်တွေ နှိပ်သော GUI အပါအဝင်၊ စာသားထည့်သွင်းခြင်း၊ ရွေးချယ်မှုတို့ ပါရှိသည်။

### Console နဲ့ UI app များက ကန့်သတ်ထားသည်

command-based app မှာ command ရိုက်တဲ့အခါ ထို app မှထောက်ခံသည့် command များသာ အသုံးပြုနိုင်တယ်။ 

- **ကန့်သတ်ထားသည်။** မည်သည့် command မဆို ရိုက်ထည့်၍ မရ၊ app ထောက်ခံသည့် command များသာ အသုံးပြုနိုင်သည်။
- **ဘာသာစကား အထူးသတ်မှတ်ထားသော။** အချို့ app များမှာ ဘာသာစကားများစွာကို ထောက်ခံနိုင်သော်လည်း ပုံမှန်အားဖြင့် တချို့ ဘာသာစကားအထူးသတ်မှတ်ထားပြီးဖြစ်သည်။ 

### စာသား ဖန်တီးအက်ပ်များ အကျိုးကျေးဇူးများ

စာသား ဖန်တီးအက်ပ်က ဘယ်လိုကွဲပြားခြားနားတာလဲ?

စာသား ဖန်တီးအက်ပ်မှာ ထောက်ခံ command များ သို့မဟုတ် နယ်ပယ်အတိအကျသော ဘာသာစကားတစ်ခုသာမက သဘာဝဘာသာစကားကို အသုံးပြု၍ ဆက်သွယ်နိုင်သည်။ ထို့ပြင် မြောက်မြားစွာ သတင်းအချက်အလက်များ ဖြတ်သန်းလေ့လာပြီး ပြန်လည်သင်ကြားထားသော ဒေတာအရင်းမြစ်နှင့် ကိုက်ညီသောကြောင့် ယာယီ app တစ်ခုလောက် Database ထဲမှာ ရှိသည့် နည်းနည်းသော အချက်အလက်ကနေလည်း ကန့်သတ်မှုမရှိပါ။

### စာသား ဖန်တီးအက်ပ်ဖြင့် မည်သည့်အရာများ ဖန်တီးနိုင်ပါသလဲ?

တော်တော်များများ ဖန်တီးနိုင်ပါတယ်။ ဥပမာ-

- **chatbot**: ကဏ္ဍများအကြောင်း၊ ကုမ္ပဏီနှင့် ထုတ်ကုန်များအကြောင်း မေးခွန်းများဖြေကြားပေးနိုင်သော chatbot တစ်ခုအကြောင်း။
- **အကူအညီပေးသူ**: LLM များသည် စာသားကို နိဂုံးချုပ်ခြင်း၊ သတင်းအချက်အလက်ယူခြင်း၊ ရုပ်သံအကြောင်းအရာများ ထုတ်လုပ်ခြင်း တို့တွင် ထူးခြားသည်။
- **ကုဒ်အကူအညီပေးသူ**: သုံးသည့် ဘာသာစကားမော်ဒယ်ပေါ်မူတည်ပြီး ကုဒ်ရေးရာတွင် ကူညီပေးနိုင်သော ကုဒ်အကူအညီပေးသူ တစ်ခု ဖန်တီးနိုင်သည်။ ဥပမာ GitHub Copilot သို့မဟုတ် ChatGPT ကွန်ယက်ကို အသုံးပြု၍ ကုဒ်ရေးနိုင်သည်။

## မည်သို့ စတင်ရမည်နည်း?

အဓိကအားဖြင့် LLM နဲ့ ပေါင်းစည်းရန် နည်းလမ်း နှစ်ခုရှိတတ်သည်-

- API အသုံးပြုခြင်း။ prompt ဖြင့် ဝဘ်လွှာ တောင်းဆိုချက်များ တည်ဆောက်ပြီး ထုတ်ပေးသော စာသားအား ပြန်လည်ဖမ်းယူပါ။
- စာကြည့်တိုက် အသုံးပြုခြင်း။ စာကြည့်တိုက်များသည် API ဆဲလ်များကို ပုန်အစကူးပြောင်းပြီး သုံးရန် ပိုလွယ်ကူစေပါသည်။

## စာကြည့်တိုက်များ/SDK များ

LLM များနှင့် အလုပ်လုပ်ရန် လူသိများသော စာကြည့်တိုက်အချို့မှာ-

- **openai**: မော်ဒယ်နှင့် ပေါင်းသင်းပြီး prompt များပို့ပေးရန် အလွယ်တကူ ချိတ်ဆက်နိုင်သော စာကြည့်တိုက်။

ထိုနောက် ထိုအထက်က အခြား အဆင့်မြင့်စာကြည့်တိုက်များလည်းရှိသည်။

- **Langchain**: Langchain တော်တော်လေး လူသိပြီး Python ကို ထောက်ခံသည်။
- **Semantic Kernel**: Semantic Kernel သည် Microsoft ၏ စာကြည့်တိုက်ဖြစ်ပြီး C#, Python, Java တို့ကို ထောက်ခံသည်။

## openai ကို အသုံးပြုသော ပထမဆုံးအက်ပ်

ပထမဆုံး အက်ပ်တစ်ခု တည်ဆောက်မည့်နည်းလမ်း၊ လိုအပ်ချက်များနှင့် တခြားအချက်များကို ကြည့်လိုက်ပါမယ်။

### openai ကို တပ်ဆင်ခြင်း

  > [!NOTE] ဒီအဆင့်ကို Codespaces သို့မဟုတ် Devcontainer အတွင်း run မလုပ်လျှင်သာ လိုအပ်သည်


OpenAI သို့မဟုတ် Azure OpenAI နှင့် ဆက်သွယ်ရန် စာကြည့်တိုက်များ အများကြီးရှိသည်။ C#, Python, JavaScript, Java နှင့် အခြား programming language များကိုလည်း အသုံးပြုနိုင်ပါသည်။  
ကျွန်ုပ်တို့က `openai` Python စာကြည့်တိုက်ကို ရွေးချယ်ပြီး `pip` ဖြင့် တပ်ဆင်ပါမည်။

```bash
pip install openai
```

Codespaces သို့မဟုတ် Dev Container အောက်တွင် run မလုပ်ဘူးဆိုရင် မိမိစက်တွင် [Python](https://www.python.org/) ကိုလည်း တပ်ဆင်ထားဖို့ လိုအပ်ပါသည်။

### resource တစ်ခု ဖန်တီးပြီး API key ရှာဖွေခြင်း

မသိရှိသေးပါက အောက်ပါအဆင့်များကို လုပ်ဆောင်ရမည်-

- OpenAI တွင် အကောင့်တစ်ခု ဖန်တီးပါ <https://platform.openai.com/signup>။
- ယခု API key များ ရယူပါ <https://platform.openai.com/api-keys>။

>[!NOTE]
> API key ကို မိမိကုဒ်နှင့် ခွဲထားဖို့ကောင်းသည်။ environment variable ကို အသုံးပြုပြီး လုပ်နိုင်ပါသည်။
> - သင်၏ .env ဖိုင်တွင် environment variable `OPENAI_KEY` ကို သင်၏ API key ဖြင့် ဖြည့်စွက်ပါ။ ဒီသင်ခန်းစာ၏ အစီရင်ခံစာများကိုပြီးမြောက်စဉ်လည်း ဒီနည်းလမ်းဖြင့် ပြင်ဆင်ထားဖြစ်နေသည်။
> - API key ကို တစ်ကြိမ်သာ အသုံးပြုနိုင်သောကြောင့် စာရိုက်မှတ်သားမှန်ကန်မှုကို သေချာစစ်ဆေးပါ။ မလည်ပတ်ပါက API key ဖြုတ်ပြီး အသစ်ပြန်ဖန်တီးရန် အကြံပြုသည်။


### OpenAI configuration ကို ပြင်ဆင်ခြင်း

OpenAI ကို အသုံးပြုနေလျှင် configuration ပြင်ဆင်နည်းမှာ အောက်ပါအတိုင်းဖြစ်သည်-

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

အထက်မှာ ကျွန်ုပ်တို့ သတ်မှတ်ထားတာများမှာ -

- `api_key`, သင်၏ OpenAI dashboard မှ ရရှိသော API key ဖြစ်သည်။
- `deployment`, သင်၏ GPT ဗားရှင်း ဖြစ်သည်။

> [!NOTE]
> `os.environ` သည် environment variable များကို ဖတ်ရန် အသုံးပြုသော function ဖြစ်သည်။ `OPENAI_API_KEY` ကဲ့သို့သော environment variable များ ဖတ်ယူနိုင်ပါသည်။

## စာသား ဖန်တီးခြင်း

စာသား ဖန်တီးရန် `responses.create` method ကို အသုံးပြုသည်။ ဥပမာကို ကြည့်ပါ-

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

အထက်ပါကုဒ်မှာ response object တစ်ခု ဖန်တီးပြီး မော်ဒယ်နှင့် prompt ကို ထည့်သွင်းသည်။ ထို့နောက် ဖန်တီးပြီးသော စာသားကို ပရင့်ထုတ်သည်။

### Chat ဖြေကြားမှုများ

ယခုအထိ ထုတ်ပြန်ထားသော prompt တစ်ခုမှ စာသားဖန်တီးနည်းကို မြင်ဖူးပြီးဖြစ်သည်။ Responses API သည် chatbots များအတွက်လည်း သင့်တော်ပြီး စကားပြော ဆက်သွယ်မှု တည်ဆောက်ရန် message စာရင်းကို input အဖြစ် ပေးနိုင်သည်။ ဥပမာကို ကြည့်ပါ-

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

ဒီလုပ်ဆောင်ချက်အကြောင်းကို နောက်ပိုင်းအခန်းတွင် အသေးစိတ် ဆက်လက်လေ့လာပါမည်။

  > [!NOTE] Codespaces သို့မဟုတ် Devcontainer အတွင်း၌ဤ notebook ကိုပြေးနိုင်ပါက ဤခြေလှမ်းမလိုအပ်ပါ။

> [!NOTE]
> သင့်ရဲ့ OpenAI key ကို https://platform.openai.com/settings/organization/api-keys သို့ သွားပြီး `API keys` ကို ရှာဖွေပါ။ အဲဒီမှာ အဲကီးအသစ်တစ်ခု ဖန်တီးနိုင်ပြီး တန်ဖိုးကို ချက်ချင်းကူးယူနိုင်ပါတယ်။


၁။ *app.py* ဖိုင်တစ်ခု ဖန်တီးပြီး အောက်ပါကုဒ်ကို ထည့်ပါ။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## အမျိုးမျိုးသော အဆိုပြုချက်များ၊ အမျိုးမျိုးသော အရာများအတွက်

အခု သင်သည် အဆိုပြုချက်ကို အသုံးပြု၍ စာသား ဖန်တီးနည်းကို ကြည့်ပြီးဖြစ်ပါပြီ။ သင်မှာ အမျိုးမျိုးသော စာသားများ ဖန်တီးနိုင်ရန် တည်းဖြတ်ပြောင်းလဲနိုင်သော အစီအစဉ်တစ်ရပ်လည်း ရှိနေပါပြီ။

အဆိုပြုချက်များကို အမျိုးမျိုးသော အလုပ်များအတွက် အသုံးပြုနိုင်သည်။ ဥပမာ-

- **စာသားအမျိုးအစားတစ်မျိုး ဖန်တီးပါ**။ ဥပမာ၊ သင်သည် ကဗျာတစ်ပုဒ်၊ မေးခွန်းများ စသည်ဖြင့် ဖန်တီးနိုင်သည်။
- **သတင်းအချက်အလက် ရှာဖွေပါ**။ 'ဝက်ဘ်ဖွံ့ဖြိုးတိုးတက်မှုတွင် CORS ဆိုသည်မှာ ဘာကို ဆိုသည်နည်း?' ကဲ့သို့သော ဥပမာအတိုင်း သတင်းအချက်အလက်ရရန် အဆိုပြုချက်များကို အသုံးပြုနိုင်သည်။
- **ကုဒ်ဖန်တီးပါ**။ အီးမေးလ်များ စစ်ဆေးရန် အသုံးပြုသောပုံစံ(regex) တစ်ခု ဖန်တီးခြင်း၊ ဒါမှမဟုတ် ဝက်ဘ်အပ်(Program) တစ်ခုလုံး ဖန်တီးခြင်း စသည်ဖြင့် ကုဒ်ဖန်တီးရန် အဆိုပြုချက်များကို အသုံးပြုနိုင်သည်။

## လေ့ကျင့်ခန်း - အစားအစာဖန်တီးသူ တည်ဆောက်ခြင်း

အခုတော့ သရုပ်ပြထားတဲ့ ဖြစ်ရပ်မှန်ကို ကစားပြီးသားဖြစ်တော့၊ ဒီ ဖြစ်ရပ်မှန်နဲ့ ကိုက်ညီတဲ့ ကုဒ်ကို ရေးသားကြပါစို့။ ဒါလုပ်ဖို့ အောက်ပါ အဆင့်တွေကို လိုက်နာပါ။

1. အရင်မှ ရှိပြီးသား *app.py* ဖိုင်ကို စတင်အနေနဲ့ သုံးပါ
1. `prompt` မတည့်ဘဲရှိနေတဲ့ variable ကို ရှာပြီး အောက်ပါကုဒ်သို့ ပြောင်းလိုက်ပါ။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> မှတ်ချက်၊ သင့် LLM သည် nondeterministic ဖြစ်သည်၊ ထို့ကြောင့် မည်သည့်အချိန်တွင်မဆို ပရိုဂရမ်ကို ခေါ်မည်ဆိုလျှင် သင့်တော်သော ရလဒ် မတူကွဲပြားနိုင်ပါသည်။

ကောင်းပါပြီ၊ ဘာတွေကို အဆင့်တိုးနိုင်ကြောင်းကြည့်ကြရအောင်။ အဆင့်တိုးရန်အတွက် ကုဒ်မှာ flexible ဖြစ်ရမည်။ အားစည်းပစ္စည်းများ နှင့် နှုန်းပေးချက်များကို ပြင်ဆင်နှင့် တိုးတက်မှု ပြုလုပ်နိုင်ရန်လိုအပ်ပါသည်။


၁။ အောက်ပါအတိုင်း ကုဒ်ကိုပြောင်းလဲကြမယ်။


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### စစ်ထုတ်ခြင်းနှင့် ဈေးဝယ်စာရင်းထည့်ခြင်းဖြင့် တိုးတက်အောင်လုပ်ခြင်း

ယခုအခါ ကျွန်ုပ်တို့တွင် လက်ရှိ အသုံးပြုသူထံမှ ဝင်ရောက်မှုများပေါ် မူတည်ပြီး ချက်ပြုတ်နည်းများ ထုတ်လုပ်နိုင်သော လုပ်ဆောင်ချက်ရှိပြီး လွယ်ကူသောအက်ပ်တစ်ခု ရှိပါသည်။ ယင်းတွင် ချက်ပြုတ်နည်း၏ အရေအတွက် နှင့် ပါဝင်ပစ္စည်းများကိုလည်း ပါဝင်သည်။

ထိုကဲ့သို့ တိုးတက်မှုအတွက် အောက်ပါအချက်များကို ထည့်သွင်းလိုပါသည်-

- **ပါဝင်ပစ္စည်း မကြိုက်နှိမ့်ခြင်းကို စစ်ထုတ်ခြင်း။** မကြိုက်ဘဲ ဖြစ်သော သို့မဟုတ် အာလ်လားဂျီ ဖြစ်သော ပါဝင်ပစ္စည်းများကို စစ်ထုတ်နိုင်စေရန်လိုသည်။ ယင်းပြောင်းလဲမှုအတွက် ကျွန်ုပ်တို့အင်ရှိ prompt ကို ပြင်ဆင်၍ အဆုံးတွင် filter အခြေအနေ တစ်ခု ထည့်သွင်းနိုင်ပါသည်-

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    မြင်သလို (milk) ပါဝင်သော ချက်ပြုတ်နည်းများအားလုံး ထုတ်ပယ်ထားသည်။ ဒါပေမဲ့ ဒါကို သောက်လို့ မရသူများအတွက် (cheese) ပါဝင်သော ချက်ပြုတ်နည်းများကိုလည်း ထုတ်ပယ်လိုသောကြောင့် လမ်းညွှန်ချက်ကို ပိုရှင်းလင်းဖွယ်ရှိပါသည်။

- **ဈေးဝယ်စာရင်း ထုတ်လုပ်ခြင်း။** အိမ်တွင် ရှိပြီးသားပစ္စည်းများကို ထည့်သွင်းစဉ်းစား၍ ဈေးဝယ်စာရင်း ထုတ်လုပ်လိုပါသည်။

    ဤလုပ်ဆောင်ချက်အတွက် တစ်ခုတည်း prompt မှ ဖြေရှင်းနိုင်သော်လည်း prompt နှစ်ခုပိုင်းခွဲ၍ လုပ်ဆောင်နိုင်သည်။ နောက်ဟာကို စမ်းကြည့်မယ်။ ထိုအတွက် ပထမ prompt ၏ ရလဒ်ကို ဒုတိယ prompt ၏ context အဖြစ် ထည့်သွင်းရန်လိုအပ်ပါသည်။

    ပထမ prompt ၏ ရလဒ်ကို ပုံနှိပ်နေသော အပိုင်းကို ရှာ၍ အောက်ပါကုဒ်ကို ထည့်သွင်းပါ-
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # ဖြေကြားချက်ကို ပုံနှိပ်ပါ
    print("Shopping list:")
    print(response.output_text)
    ```

    အောက်ပါအချက်များကို မှတ်သားပါ-

    - ပထမ prompt ၏ ရလဒ်ကို ဒုတိယ prompt မှာ ထည့်သွင်းပြီး prompt အသစ်တစ်ခု ဖန်တီးပါသည်-
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - အသစ်တစ်ခုတောင်းဆိုမှု ဖြစ်အောင် ပြုလုပ်ပြီး ပထမ prompt တွင် တောင်းခဲ့သော token အရေအတွက်ကိုလည်း စဉ်းစားထားပါသည်၊ ရှေ့နေရာတွင် `max_output_tokens` ကို 1200 သတ်မှတ်ထားသည်။

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        ဒီကုဒ်ကို စမ်းသပ်သုံး ကြည့်ပါက အောက်ပါ အမှတ်အသားများထုတ်ဖေါ်ပါသည်-

        ```output
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice
        ```

- **token ရဲ့အရှည်အကြောင်း**။ ကျွန်ုပ်တို့လိုအပ်သော စာသားကို ထုတ်ဖော်ရန် token အရေအတွက်ကို စဉ်းစားသင့်သည်။ token သည် ငွေကြေးကုန်ကျစရိတ်ဖြစ်သောကြောင့် မဖြစ်မနေ token ကို စွမ်းဆောင်မှုအရ အကျိုးရှိစွာအသုံးပြုသင့်သည်။ ဥပမာအားဖြင့် prompt ကို token များနည်းအောင် ဖော်ပြနိုင်သလား စဉ်းစားပါ။

   အသုံးပြု token အရေအတွက်ကို ပြောင်းလဲရန် `max_output_tokens` ပရမီတာကို အသုံးပြုနိုင်သည်။ ဥပမာ token ၁၀၀ သုံးချင်ပါက-

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **အပူချိန်နှင့် စမ်းသပ်ခြင်း**။ အပူချိန်သည် မြင်သာစွာရှင်းပြထားခြင်း မရှိသေးပေမဲ့ ကျွန်ုပ်တို့၏ အစီအစဉ်မည်သို့ လုပ်ဆောင်မည်ကို သက်ဆိုင်သော ဆက်စပ်အချက်ဖြစ်သည်။ အပူချိန်တန်ဖိုးမြင့်မားလျှင် ထွက်ရှိသောရလဒ်သည် ပိုမိုသတိမပြုရွာနေပါသည်။ အပူချိန်တန်ဖိုးနိမ့်လျှင် ထွက်ရှိသော ရလဒ်သည် ပိုမိုခန့်မှန်းနိုင်သည်။ ထွက်ရှိမှုတွင် မတူညီမှုလိုချင်သည်၊ မလိုချင်သို့ မဖြစ်ရန် စဉ်းစားပါ။

   အပူချိန်ကို ပြောင်းလဲရန် `temperature` parameter ကို အသုံးပြုနိုင်သည်။ ဥပမာ အပူချိန် 0.5 သတ်မှတ်လိုပါက အောက်ပါနည်းဖြင့်လုပ်နိုင်သည်-

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```
