# စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းများ ဖန်တီးခြင်း

ဒီသင်ရိုးအစီအစဉ်မှတဆင့် သင် မြင်တွေ့ရသောအတိုင်း prompt များနှင့် "prompt engineering" ဟူသော ဘာသာရပ်တစ်ခုလုံးကဲ့သို့သော အဓိကသဘောတရားများ ရှိသည်။ ChatGPT, Office 365, Microsoft Power Platform အပါအဝင် သင့်အား prompt များကို အသုံးပြုပြီး အရာတစ်ခုခု ပြုလုပ်နိုင်သော ကိရိယာများစွာရှိသည်။

အဲ့ဒီလိုအတွေ့အကြုံတစ်ခုကို အက်ပလီကေးရှင်းတစ်ခုတွင် ထည့်သွင်းချင်ပါက prompt, completion ကဲ့သို့သော သဘောတရားများနားလည်ရမည်ဖြစ်ပြီး အသုံးပြုရန်လိုသော ကတ်ဖတ်တစ်ခုကို ရွေးချယ်ရမည်။ ဒါပေမဲ့ ဒီအခန်းမှာ သင် သင်ယူသွားမည်ဖြစ်သည်။

## နိဒါန်း

ဒီအခန်းတွင် သင်: 

- openai ကတ်ဖတ် နှင့် ၎င်း၏ အဓိကသဘောတရားများကို လေ့လာမည်။
- openai ကို အသုံးပြု၍ စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းတစ်ခု တည်ဆောက်မည်။
- prompt, temperature, token များကဲ့သို့သော သဘောတရားများကို အသုံးပြုပြီး စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းတစ်ခု ဖန်တီးနည်းကို နားလည်မည်။

## သင်ယူရန်ရည်မှန်းချက်များ

ဒီသင်ခန်းစာ၏အဆုံးတွင် သင်သည်:

- စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းသည် အဘယ်ဟု ဆိုသည်ကို ရှင်းပြတတ်မည်။
- openai ကို အသုံးပြု၍ စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းတစ်ခု တည်ဆောက်မည်။
- သင့်အက်ပလီကေးရှင်းတွင် token များ ပိုများ သို့မဟုတ် နည်းလာစေရန် ပြင်ဆင်ခြင်း၊ temperature ကို လည်း ပြောင်းလဲ၍ အမျိုးမျိုးသော output များ ရရှိစေမည်။

## စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်း ဆိုသည်မှာ အဘယ်နည်း?

ပုံမှန်အားဖြင့် အက်ပလီကေးရှင်းတစ်ခု တည်ဆောက်သည်မှာ အောက်ပါပုံစံအတိုင်း အင်တာဖေ့စ်တစ်ခုရှိသည်။

- Command-based။ ကန်ဆိုလ်အက်ပလီကေးရှင်းများသည် သင် command တစ်ခု ရိုက်ထည့်ပြီး တာဝန်တစ်ခု ပြုလုပ်သည့် တစ်မျိုးအက်ပလီကေးရှင်းများဖြစ်သည်။ ဥပမာအားဖြင့် `git` သည် command-based app တစ်ခုဖြစ်သည်။
- User interface (UI)။ အချို့အက်ပလီကေးရှင်းများတွင် graphical user interfaces (GUIs) ရှိပြီး သင် ခလုတ်များကိုနှိပ်၊ စာသားထည့်၊ ရွေးချယ်မှုများပြုလုပ်နိုင်သည်။

### ကန်ဆိုလ်နှင့် UI အက်ပလီကေးရှင်းများသည် ကန့်သတ်ချက်ရှိသည်

Command-based app တစ်ခု သုံးသည့်အခါ ရိုက်ထည့်သော command များကို နှိုင်းယှဉ်ကြည့်ပါ -

- **ကန့်သတ်ထားသည်။** အက်ပလီကေးရှင်းထောက်ပံ့သည့် command များသာ ရိုက်ထည့်နိုင်သည်၊ မည်သည့် command မဆို မရိုက်နိုင်ပါ။
- **ဘာသာစကားအထူးပြု။** အချို့အက်ပလီကေးရှင်းတွင် ဘာသာစကားများစွာ အထောက်အပံ့ရှိသော်လည်း ပုံမှန်အားဖြင့် သီးသန့်ဘာသာစကားအတွက် ဖန်တီးထားသည်၊ ထိုသို့ပင် ဘာသာစကားထပ်ထည့်ရန်ဖြစ်ပါကဖြစ်။

### စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်း၏ အကျိုးကျေးဇူးများ

စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းကဘာနည်း?

စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းတွင် သင်သည် ပို၍ချိုးဖောက်နိုင်ပြီး command များ တစ်စုသတ်မှတ်ခြင်းမရှိ၊ အထူးသတ်မှတ်ဘာသာစကားသတ်မှတ်ခြင်းမရှိပဲ ရောသဘာဝဘာသာစကားဖြင့် အက်ပလီကေးရှင်းနှင့် ဆက်သွယ်နိုင်သည်။ ထို့အပြင် သင် ဆက်သွယ်နေသော ဒေတာအရင်းအမြစ်သည် အကြီးစား သတင်းအချက်အလက်များ ပေါင်းစုံ၌ လေ့ကျင့်ထားသော ထောက်ပံ့မှုရှိသောကြောင့်၊ ရိုးရိုးအက်ပလီကေးရှင်းအတွက် database တွင်ထားရှိသော အချက်အလက်များသာ ယုံကြည်ရနိုင်သည်။

### စာသားဖန်တီးခြင်းအက်ပလီကေးရှင်းဖြင့် ဘာတွေ ဖန်တီးနိုင်မလဲ?

ပြုလုပ်နိုင်သောအရာများစွာရှိသည်။ ဥပမာအားဖြင့် -

- **Chatbot တစ်ခု** ။ သင်၏ကုမ္ပဏီနှင့် ထုတ်ကုန်များအကြောင်း ဆွေးနွေးမေးမြန်းနိုင်သော chatbot တစ်ခု ဖြစ်စေမည်။
- **အကူအညီပေးသူ** ။ LLM များသည် စာသားအကျဥ်းချုပ်ခြင်း၊ စာသားမှ မှတ်ယူချက်ရယူခြင်း၊ နောက်ထပ် resume အပြီးသတ် စာသားဖန်တီးခြင်းတို့တွင် အကောင်းဆုံး ဖြစ်သည်။
- **Code assistant** ။ သုံးသည့် ဘာသာစကား မော်ဒယ်ပေါ် မူတည်၍ code ရေးရာတွင် အကူအညီပေးနိုင်သော code assistant တစ်ခု ဖန်တီးနိုင်သည်။ ဥပမာအနေနှင့် GitHub Copilot သို့မဟုတ် ChatGPT ကို အသုံးပြု၍ ကိုးဒ်ရေးခြင်း အကူအညီ ရနိုင်သည်။

## စတင်ဘယ်လိုလုပ်မလဲ?

LLM တစ်ခုနှင့် ပေါင်းစည်းရန် နည်းလမ်း ၂ မျိုး ရှိသည် -

- API အသုံးပြုခြင်း။ ဤနေရာတွင် သင်၏ prompt ဖြင့် ဝက်ဘ်တောင်းဆိုမှုများ ပြုလုပ်ပြီး ဖန်တီးမည့် စာသားကို ပြန်လည်ရရှိပါမည်။
- ကတ်ဖတ်အသုံးပြုခြင်း။ ကတ်ဖတ်များသည် API ခေါ်ဆိုမှုများကို အုပ်ချုပ်ကာ အသုံးပြုရ လွယ်ကူစေသည်။

## ကတ်ဖတ်များ/SDK များ

LLM များတွင် အသုံးပြုရန် နာမည်ကြီး ကတ်ဖတ်များ ရှိသည်၊ ဥပမာ -

- **openai**, ဤကတ်ဖတ်သည် မော်ဒယ်နှင့် ချိတ်ဆက်နိုင်ခြင်းနှင့် prompt ရိုက်ထည့်ခြင်းကို လွယ်ကူစေသည်။

ထို့အပြင် အဆင့်မြင့် ကတ်ဖတ်များ ပါရှိသည် -

- **Langchain**। Langchain သည် နာမည်ကြီးပြီး Python ကိုထောက်ပံ့သည်။
- **Semantic Kernel**။ Semantic Kernel သည် Microsoft ၏ ကတ်ဖတ်ဖြစ်ပြီး C#, Python, Java ဘာသာစကားများကို ထောက်ပံ့သည်။

## လေ့ကျင့်ခန်း - သူ့နှင့်ကိုက်ညီသောချက်ပြုတ်နည်း များဖန်တီးသူတစ်ဦး

အခု Scenario တစ်ခုကို ပြသပြီးစီးသွားပါပြီ၊ ပြသခဲ့သော Scenario နှင့်ကိုက်ညီသည့် ကုဒ်များရေးသားလိုက်ကြစို့။ ဒီလိုလုပ်ရန်အတွက် အောက်ပါအဆင့်များအား လိုက်ပါရန်လိုသည်။

1. ရှိပြီးသားဖိုင်ကို စတင်အနေဖြင့်အသုံးပြုပါ
1. `prompt` variable ကိုဖန်တီးပြီး နမူနာကုဒ်ကို အောက်ပါအတိုင်း ပြောင်းလဲပါ။


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

အခုကိုဒ်ကို run လုပ်ခဲ့ရင် ဒီလို output တစ်ခုကိုတွေ့မှာဖြစ်ပါတယ်။

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
