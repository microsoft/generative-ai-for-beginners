# စာသားထုတ်လုပ်တဲ့အက်ပ်တွေ တည်ဆောက်ခြင်း

ဒီသင်တန်းတစ်လျှောက်လုံးမှာ သင်မြင်တွေ့ခဲ့တာက prompt တွေလို အဓိကအယူအဆတွေနဲ့ "prompt engineering" ဆိုတဲ့ အထူးအတန်းတစ်ခုတောင် ရှိနေတယ်ဆိုတာပါ။ ChatGPT, Office 365, Microsoft Power Platform တို့လို သုံးနိုင်တဲ့ကိရိယာတွေမှာလည်း prompt တွေကို အသုံးပြုပြီး တစ်ခုခုလုပ်ဆောင်နိုင်အောင် ထောက်ပံ့ပေးထားပါတယ်။

သင့်အက်ပ်မှာ ဒီလိုအတွေ့အကြုံတစ်ခု ထည့်သွင်းချင်ရင်တော့ prompt, completions တို့လို အယူအဆတွေနဲ့ အသုံးပြုမယ့် library ကိုရွေးချယ်တတ်ဖို့ လိုအပ်ပါတယ်။ ဒီအခန်းမှာတော့ ဒီအကြောင်းတွေကို သင်လေ့လာရမှာပါ။

## မိတ်ဆက်

ဒီအခန်းမှာ သင်

- openai library နဲ့ အဓိကအယူအဆတွေကို လေ့လာရမယ်။
- openai ကိုသုံးပြီး စာသားထုတ်လုပ်တဲ့အက်ပ်တစ်ခု တည်ဆောက်ရမယ်။
- prompt, temperature, tokens တို့လို အယူအဆတွေကို အသုံးပြုပြီး စာသားထုတ်လုပ်တဲ့အက်ပ်တစ်ခု ဘယ်လိုတည်ဆောက်ရမလဲ ဆိုတာ နားလည်လာမယ်။

## သင်ယူရမယ့်ရည်မှန်းချက်များ

ဒီသင်ခန်းစာအဆုံးမှာ သင်

- စာသားထုတ်လုပ်တဲ့အက်ပ်ဆိုတာ ဘာလဲ ဆိုတာ ရှင်းပြနိုင်မယ်။
- openai ကိုသုံးပြီး စာသားထုတ်လုပ်တဲ့အက်ပ်တစ်ခု တည်ဆောက်နိုင်မယ်။
- သင့်အက်ပ်မှာ tokens များ/နည်း အသုံးပြုနိုင်သလို, temperature ကိုလည်း ပြောင်းလဲနိုင်မယ်။ ဒါကြောင့် output ကို မတူညီအောင် ထုတ်နိုင်မယ်။

## စာသားထုတ်လုပ်တဲ့အက်ပ်ဆိုတာဘာလဲ?

ပုံမှန်အက်ပ်တစ်ခု တည်ဆောက်တဲ့အခါမှာ အောက်ပါအတိုင်း interface တစ်ခုခု ရှိတတ်ပါတယ်-

- Command-based။ Console app တွေက သင် command တစ်ခု ရိုက်ထည့်ပြီး တစ်ခုခုလုပ်ဆောင်တဲ့ app တွေပါ။ ဥပမာ `git` က command-based app တစ်ခုပါ။
- User interface (UI)။ အချို့အက်ပ်တွေမှာ graphical user interface (GUI) ရှိပြီး button တွေကိုနှိပ်တာ, စာရိုက်ထည့်တာ, ရွေးချယ်မှုတွေပြုလုပ်တာ စသဖြင့် လုပ်နိုင်ပါတယ်။

### Console နဲ့ UI အက်ပ်တွေက အကန့်အသတ်ရှိတယ်

Command-based app တစ်ခုမှာ command ရိုက်ထည့်တဲ့အခါ

- **အကန့်အသတ်ရှိတယ်**။ သင် app ကထောက်ပံ့တဲ့ command တွေကိုသာ ရိုက်ထည့်နိုင်ပြီး, မည်သည့် command မဆို ရိုက်လို့မရပါ။
- **ဘာသာစကားအထူးပြု**။ အချို့အက်ပ်တွေက ဘာသာစကားအမျိုးမျိုးကို ထောက်ပံ့ပေမယ့်, မူလအနေနဲ့ တစ်ခုတည်းသော ဘာသာစကားအတွက် တည်ဆောက်ထားတာများပါတယ်။ နောက်ထပ်ဘာသာစကားထည့်နိုင်ပေမယ့်ပါ။

### စာသားထုတ်လုပ်တဲ့အက်ပ်တွေရဲ့ အားသာချက်များ

ဒါဆို စာသားထုတ်လုပ်တဲ့အက်ပ်က ဘာကြောင့် မတူသလဲ?

စာသားထုတ်လုပ်တဲ့အက်ပ်မှာ သင်ပိုပြီး လွတ်လပ်မှုရှိတယ်။ သတ်မှတ်ထားတဲ့ command တွေ၊ input ဘာသာစကား တစ်ခုတည်းနဲ့ မကန့်သတ်ဘူး။ သာမန်ဘာသာစကားနဲ့ app နဲ့ ဆက်သွယ်နိုင်တယ်။ နောက်ထပ် အားသာချက်ကတော့ သင်သုံးနေတဲ့ data source က အကြီးအကျယ်သင်ကြားထားတဲ့ အချက်အလက်တွေပါဝင်လို့, သာမန် app တစ်ခုလို database ထဲက အချက်အလက်နဲ့သာ ကန့်သတ်မနေဘူး။

### စာသားထုတ်လုပ်တဲ့အက်ပ်နဲ့ ဘာတွေတည်ဆောက်နိုင်လဲ?

တည်ဆောက်နိုင်တာတွေ များစွာရှိတယ်။ ဥပမာ-

- **Chatbot**။ သင့်ကုမ္ပဏီနဲ့ ထုတ်ကုန်တွေကို မေးခွန်းဖြေတဲ့ chatbot တစ်ခု တည်ဆောက်နိုင်တယ်။
- **Helper**။ LLM တွေက စာသားအကျဉ်းချုပ်ရေးခြင်း, အချက်အလက်ထုတ်ယူခြင်း, CV တို့လို စာသားထုတ်လုပ်ခြင်းမှာ အထူးကောင်းပါတယ်။
- **Code assistant**။ သုံးမယ့် language model ပေါ်မူတည်ပြီး, သင့်ကို code ရေးရာမှာ ကူညီပေးနိုင်တဲ့ code assistant တစ်ခုတည်ဆောက်နိုင်တယ်။ ဥပမာ GitHub Copilot, ChatGPT တို့ကို code ရေးရာမှာ အသုံးပြုနိုင်တယ်။

## ဘယ်လိုစတင်ရမလဲ?

LLM နဲ့ ပေါင်းစည်းဖို့ နည်းလမ်းနှစ်ခုရှိတယ်-

- API အသုံးပြုခြင်း။ ဒီမှာတော့ သင့် prompt နဲ့ web request တစ်ခု တည်ဆောက်ပြီး, ပြန်လာတဲ့ စာသားကို အသုံးပြုရတယ်။
- Library အသုံးပြုခြင်း။ Library တွေက API ခေါ်တာတွေကို လွယ်ကူအောင် encapsulate လုပ်ပေးတယ်။

## Libraries/SDKs

LLM တွေနဲ့ အလုပ်လုပ်ဖို့ နာမည်ကြီး library တွေရှိတယ်-

- **openai** - ဒီ library က သင့် model နဲ့ ချိတ်ဆက်ပြီး prompt တွေ ပို့ဖို့ လွယ်ကူစေတယ်။

ထို့အပြင်, ပိုမြင့်မားတဲ့အဆင့်မှာ အလုပ်လုပ်တဲ့ library တွေလည်းရှိတယ်-

- **Langchain**။ Langchain က နာမည်ကြီးပြီး Python ကို ထောက်ပံ့တယ်။
- **Semantic Kernel**။ Semantic Kernel က Microsoft မှ ထုတ်လုပ်ထားပြီး C#, Python, Java တို့ကို ထောက်ပံ့တယ်။

### 1. Codespaces 

* 'Get started' window မှာ "Run codespace" ကို ရွေးပါ။
* codespace အသစ်တစ်ခု ဖန်တီးပါ (သို့မဟုတ် ရှိပြီးသားကို အသုံးပြုပါ)
* VS Code က သင့် browser မှာ ဖွင့်ပေးပြီး, စမ်းသပ်နိုင်တဲ့ notebook များစွာကို ဘာသာစကားအမျိုးမျိုးနဲ့ မြင်ရပါလိမ့်မယ်။
* ```./githubmodels-app.py``` ကို run လုပ်ပါ။

> Note: codespaces မှာ Github Token variable ကို သတ်မှတ်ဖို့ မလိုအပ်ပါဘူး၊ ဒီအဆင့်ကို ကျော်သွားနိုင်ပါတယ်။

**အောက်မှာရှိတဲ့ 'Generate Text' အပိုင်းကို ဆက်လက်လုပ်ဆောင်ပါ**

### 2. VS Code (သို့မဟုတ် မိမိကြိုက်တဲ့ IDE)

'Get started' အစိမ်းရောင် button မှာ မိမိကြိုက်တဲ့ IDE မှာ run လုပ်ဖို့ လိုအပ်တဲ့ အချက်အလက်အားလုံး ရနိုင်ပါတယ်။ ဥပမာအနေနဲ့ VS Code ကို ပြပါမယ်။

* ဘာသာစကားနဲ့ SDK ကို ရွေးပါ - ဥပမာ Python နဲ့ Azure AI Inference SDK ကို ရွေးမယ်။
* GitHub မှာ personal access token တစ်ခု ဖန်တီးပါ။ ဒါကို Developer Settings မှာ ရနိုင်ပါတယ်။ ဒီ token ကို permission မပေးရပါဘူး။ သတိပြုရမှာက token ကို Microsoft service ကို ပို့မှာပါ။
* Github personal access token ကို သိမ်းထားဖို့ environment variable တစ်ခု ဖန်တီးပါ - bash, powershell, windows command prompt အတွက် နမူနာတွေ ရနိုင်ပါတယ်။

## လေ့ကျင့်ခန်း - ဟင်းချက်နည်းထုတ်ပေးစက် တည်ဆောက်ခြင်း

အခုတော့ အကြောင်းအရာတစ်ခုကို သရုပ်ပြပြီးပြီဆိုတော့၊ သရုပ်ပြထားတဲ့အခြေအနေနဲ့ ကိုက်ညီတဲ့ code ကိုရေးကြပါမယ်။ အဲ့ဒါလုပ်ဖို့အတွက် အောက်ပါအဆင့်တွေကို လိုက်နာပါ။

1. ရှိပြီးသားဖိုင်ကို အခြေခံပြီး စတင်ပါ
1. `prompt` variable တစ်ခု ဖန်တီးပြီး နမူနာ code ကို အောက်ပါအတိုင်း ပြောင်းလဲပါ:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

အခု code ကို run လုပ်ကြည့်မယ်ဆိုရင်၊ အောက်ပါ output နဲ့ဆင်တူတဲ့အရာကိုတွေ့ရပါလိမ့်မယ် -

> NOTE, သင့် LLM ဟာ nondeterministic ဖြစ်တဲ့အတွက်၊ program ကို run လုပ်တိုင်း result မတူနိုင်ပါတယ်။

အဆင်ပြေပါပြီ၊ အခုတော့ ဘယ်လိုပိုမိုကောင်းမွန်အောင်လုပ်နိုင်မလဲ ကြည့်ကြည့်ရအောင်။ ပိုမိုကောင်းမွန်အောင်လုပ်ဖို့ဆိုရင်၊ code ကို flexible ဖြစ်အောင်လုပ်ဖို့လိုပါတယ်။ ဒါကြောင့် အသုံးပြုမယ့် အစိတ်အပိုင်းတွေ၊ recipes အရေအတွက်တွေကို လိုအပ်သလို ပြင်ဆင်နိုင်အောင်လုပ်နိုင်ပါမယ်။


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

အခုတော့ အသုံးပြုနိုင်တဲ့ app တစ်ခု ရရှိပြီးသားဖြစ်ပါတယ်။ အသုံးပြုသူရဲ့ input တွေကို အခြေခံထားတဲ့အတွက် တော်တော်လည်း ပြောင်းလဲနိုင်ပါတယ်။ ဥပမာ၊ မုန့်ချက်နည်းအရေအတွက်၊ အသုံးပြုမယ့် ပစ္စည်းတွေ စသဖြင့်။
