# စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းများ ဖန်တီးခြင်း

ဒီ သင်ခန်းစာတွေအတွင်းမှာ နောက်ဆုံးအထိ တွေ့ရသလို prompt တွေ၊ prompt အင်ဂျင်နီယာနယ်လိုယ် ခေါ်တဲ့ များသောခန်းကျုတ်လည်း ရှိတယ်။ ChatGPT, Office 365, Microsoft Power Platform တို့လို အမျိုးမျိုးသောကိရိယာများဟာ prompt တွေကို အသုံးပြုပြီး တစ်စိတ်တစ်ပိုင်း ပြီးမြောက်မြေအောင် ကူညီပေးတယ်။

ဒီလို အတွေ့အကြုံကို သင့်အက်ပလီကေးရှင်းထဲ ထည့်သွင်းဖို့အတွက် prompt, completion စတဲ့ အယူအဆတွေကို နားလည်ဖို့လိုတယ်။ ဒီ ခန်းမှာတော့ သင်တို့ အကြောင်းအရာတွေကို အောက်ပါအတိုင်း သင်ယူသွားမှာ ဖြစ်တယ်။

## နိဒါန်း

ဒီ ခန်းထဲမှာ သင်လုပ်ဆောင်ရန်များမှာ -

- openai library နဲ့ အဓိက အယူအဆတွေကို သင်ယူမယ်။
- openai ကို အသုံးပြုကာ စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းတစ်ခု တည်ဆောက်မယ်။
- prompt, temperature, token စတဲ့ အယူအဆတွေကို အသုံးပြုပြီး စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းတည်ဆောက်နည်းကို နားလည်မယ်။

## သင်ယူမည့် ရည်မှန်းချက်များ

ဒီသင်ခန်းစာအဆုံးတွင် သင်သည် -

- စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းဆိုတာဘာလဲဆိုတာ ရှင်းပြနိုင်မယ်။
- openai ကို အသုံးပြုပြီး စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းတည်ဆောက်နိုင်မယ်။
- အသုံးပြုသူထည့်သွင်းသော token အရေအတွက်များကို ပိုလုပ်ရန် သို့မဟုတ် နည်းနည်း လျော့ချရန်၊ temperature ကို ပြောင်းလဲ၍ အမျိုးမျိုးသောရလဒ်များ ရနိုင်အောင် ဖန်တီးနိုင်မယ်။

## စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းဆိုတာ ဘာလဲ?

မူလတည်ဆောက်တဲ့ အက်ပလီကေးရှင်းတွေဟာ အောက်ပါကဲ့သို့ အင်တာဖေ့စ်တစ်မျိုးခုရှိတတ်တယ်။

- Command-based။ console အက်ပလီကေးရှင်းတွေက command တစ်ခု type လုပ်ပြီး တာဝန်တစ်ခု ပြီးမြောက်စေတယ်။ ဥပမာ `git` က command-based အက်ပလီကေးရှင်းတစ်ခုဖြစ်တယ်။
- User interface (UI)။ အချို့ အက်ပလီကေးရှင်းတွေမှာ graphical user interface (GUI) ရှိပြီး button တွေးကို နှိပ်ခြင်း၊ အချက်အလက်ထည့်ခြင်း၊ ရွေးချယ်ချက်များကို ကလစ်ခြင်းတို့ ပါဝင်တတ်သည်။

### Console နဲ့ UI အက်ပလီကေးရှင်းတွေက ကန့်သတ်ချက်ရှိတယ်

command-based အက်ပလီကေးရှင်းတွေနဲ့ နှိုင်းယှဉ်ကြည့်ရင် -

- **ကန့်သတ်ချက်ရှိတယ်**။ မည်သည့်commandကိုမဆို type လုပ်၍မရ၊ အက်ပလီကေးရှင်းက ထောက်ခံတဲ့ command တွေကိုသာ type တတ်ရမည်။
- **ဘာသာစကား အတိအကျဖြစ်တယ်**။ အက်ပလီကေးရှင်းများအချို့က အများအပြား ဘာသာစကားထောက်ခံချက် ရှိတယ်၊ သို့သော် ပုံမှန်အားဖြင့် အက်ပလီကေးရှင်းကို ထုတ်လုပ်တဲ့ ဘာသာစကားတစ်ခုသာ ဖြစ်ပြီး ဘာသာစကား ထောက်ခံချက်တွေ ထည့်နိုင်တယ်။

### စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းရဲ့ အားသာချက်များ

စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းက ဘာကွာခြားလဲ?

စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းမှာ မည်သည့် command မရှိပဲ သဘာဝဘာသာစကားများကို အသုံးပြုပြီး အက်ပလီကေးရှင်းနဲ့ ဆက်သွယ်နိုင်ပြီး ပိုမိုလွတ်လပ်မှု ရရှိသည်။ ထို့အပြင် သင်အချက်အလက်အရင်းမြစ်တစ်ခုနှင့် ဆက်သွယ်နေပြီး AI သင်တန်းကျောင်း extensive အချက်အလက်များကိုလည်း ထိတွေ့ရနိုင်တယ်၊ ဒါကြောင့် ပုံမှန် app ထက် ပိုမိုကျယ်ပြန့်တဲ့ အချက်အလက်အခြေခံ ထောက်ခံမှု ရရှိပါတယ်။

### စာသား ဖန်တီးရေး အက်ပလီကေးရှင်းနဲ့ ဘာတွေ ဖန်တီးနိုင်မလဲ?

ဘယ်လိုတွေ ဖန်တီးနိုင်သလဲ ရှင်းပြဖို့ -

- **Chatbot တစ်ခု**။ သင့်ကုမ္ပဏီနှင့် ထုတ်ကုန်များအပေါ် မေးခွန်းများကို ဖြေကြားပေးတဲ့ chatbot ကောင်းတစ်ခု ဖြစ်နိုင်သည်။
- **အကူအညီပေးသူ**။ LLM များက စာသားအကျဉ်းချုပ်ခြင်း၊ စာသားမှ သဘောတရား ယူဆချက်များ ရယူခြင်း၊ အတူတူ စာသားများကို ထုတ်လုပ်နိုင်ခြင်း (ဥပမာ ရုပ်သွင်ပြင်ဆောင်းပါးများ) စသည်တို့၌ စွမ်းရည်ကောင်းသည်။
- **Code assistant**။ သင် အသုံးပြုမယ့် language model အပေါ် မူတည်ပြီး၊ သင့်ကို code ရေးရာတွင် ကူညီပေးနိုင်သော code assistant တစ်ခု ဆောက်နိုင်သည်။ ဥပမာ GitHub Copilot နှင့် ChatGPT ကို code ရေးရန်ကူညီပေးရန် အသုံးပြုနိုင်သည်။

## ကနေ ထည့်သွင်းဖြစ်မလဲ?

ကောင်းပါတယ်၊ LLM တစ်ခုနှင့် ပေါင်းစည်းတိုက်ရိုက်လုပ်ဆောင်ဖို့ အောက်ပါနည်းလမ်းနှစ်ခု ပါဝင်တတ်ပါတယ်။

- API ကိုအသုံးပြုရန်။ ဤနေရာတွင် prompt ကို ထည့်သွင်းပြီး web request ဖန်တီးပြီး ပြန်လာသော စာသားကိုရယူပါ။
- Library ကိုအသုံးပြုရန်။ Library များက API ခေါ်ဆောင်မှုများကို လွယ်ကူစွာ သုံးနိုင်အောင် ထုပ်ပိုးပေးသည်။

## Libraries/SDKs

LLM များနှင့် အလုပ်လုပ်ရန် ကျော်ကြားသော library များ တချို့ရှိသည်၊ ဥပမာ -

- **openai**၊ ဤ library က သင့်မော်ဒယ်နှင့် ချိတ်ဆက်၍ prompt ပေးပို့ရလွယ်ကူစေသည်။

ထိုအပြင် အဆင့်မြင့် library များလည်း ရှိသည်၊

- **Langchain**။ Langchain ကို ကျော်ကြားပြီး Python အတွက် ထောက်ခံသည်။
- **Semantic Kernel**။ Semantic Kernel သည် Microsoft မှ ထုတ်လုပ်ပြီး C#, Python, Java ဘာသာစကားများကို ထောက်ပံ့သည်။

## Microsoft Foundry Models Playground နှင့် Azure AI Inference SDK ကို အသုံးပြုပြီး ပထမဆုံး အက်ပလီကေးရှင်းတည်ဆောက်ခြင်း

ပထမဆုံး app တည်ဆောက်နည်း၊ ဘယ် library တွေလိုအပ်မလဲ၊ အလိုအလျောက် မည်မျှလိုအပ်မလဲ အစရှိစွာကို ကြည့်ကြပါစို့။

> **မှတ်ချက်:** GitHub Models သည် ၂၀၂၆ ခုနှစ် ဇူလိုင်လကုန်တွင် ပယ်ဖျက်သွားမှာ ဖြစ်ပြီး အောက်ပါအဆင့်များကို သက်ဆိုင်ရာ [Microsoft Foundry Models](https://ai.azure.com/catalog/models?WT.mc_id=academic-105485-koreyst) သို့ ပြောင်းလဲထားပြီး GitHub Models က ပေးသော "စမ်းသပ်ရန်၊ playground တွင်ကြိုးစားရန်နှင့် API key ဖြင့် ခေါ်ရန်" အတွေ့အကြုံတုန့်ပြန်မှုကို Microsoft Foundry portal မှ တိုက်ရိုက် ပေးဆောင်နေပါသည်။ အပြည့်အဝ offline အလုပ်လုပ်ရန် ဦးစားပေးပါသလား? [Foundry Local](https://foundrylocal.ai?WT.mc_id=academic-105485-koreyst) ကို ကြည့်ပါ - cloud subscription မလိုအပ်ပဲ သင့်စက်ပေါ်မှာ models များ သယ်ဆောင်အသုံးပြုနိုင်ပါသည်။

ယခု သင်တစ်ခုသော prompt ကို စကိုင်၍ စာသား ထုတ်လုပ်နည်းကို မြင်တွေ့ပြီးသားဖြစ်သည်။ သင်မှာ ပြောင်းလဲ ပြင်ဆင်နိုင်သော အစီအစဉ်တစ်ခုလည်း ရှိပြီး အမျိုးမျိုးသော စာသားမျိုးများ ထုတ်လုပ်နိုင်ပါပြီ။

Prompt များကို အလုပ်အမျိုးမျိုးအတွက် အသုံးပြုနိုင်ပါသည်။ ဥပမာအားဖြင့် -

- **စာသား အမျိုးအစား တစ်မျိုး ထုတ်လုပ်ခြင်း**။ ဥပမာ - ကဗျာတစ်ပုဒ်၊ အစစ်အမှန် စမ်းသပ်မေးခွန်းများ ထုတ်ပြုတ်နိုင်သည်။
- **သတင်းအချက်အလက် ရှာဖွေခြင်း**။ "ဝက်ဘ်ဖွံ့ဖြိုးရေးတွင် CORS ဆိုသည်မှာ ဘာလဲ?" ကဲ့သို့ အချက်အလက် ရှာဖွေဖို့ prompt များကို အသုံးပြုနိုင်ပါသည်။

## ခြေလှမ်းလုပ်ဆောင်မှု - ဟင်းချက်မွမ်းမံရေးထုတ်လုပ်သူ တည်ဆောက်ခြင်း

ယခု ကျွန်ုပ်တို့သည် အခြေအနေတစ်ခုကို ဆက်တိုက်ပြသပြီးဖြစ်သောကြောင့်၊ အဆိုပါ ပြသထားသော အခြေအနေနှင့် ကိုက်ညီသော ကုဒ်တစ်ခုရေးလိုက်ပါစို့။ ထိုအတွက် အောက်ပါအဆင့်များကို လိုက်နာပါ။

1။ ရှိပြီးသား ဖိုင်ကို စတင်အနေဖြင့် အသုံးပြုပါ။
1။ `prompt` ပရိုဂရမ်ဗားရီးယားကို တည်ဆောက်ပြီး နမူနာကုဒ်ကို အောက်ပါအတိုင်း ပြင်ဆင်ပါ။


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
