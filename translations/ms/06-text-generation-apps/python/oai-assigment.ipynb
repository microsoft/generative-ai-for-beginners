## Jana teks

Cara untuk menjana teks adalah menggunakan kaedah `responses.create`. Berikut adalah contoh:

```python

```

Dalam kod di atas, kita mencipta objek respons dan memasukkan model yang kita mahu gunakan dan prompt. Kemudian kita cetak teks yang dijana.

### Respons sembang

Setakat ini, anda telah melihat bagaimana kita boleh menjana teks dari satu prompt. API Respons juga sangat sesuai untuk chatbot - anda menyediakan senarai mesej sebagai `input` untuk membina perbualan. Berikut adalah contoh menggunakannya:

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Di sini kami mencadangkan menambah prompt tambahan, tetapi untuk ini berfungsi, kami perlu menambah hasil daripada prompt pertama sebagai konteks kepada prompt kedua.

    Cari bahagian dalam kod yang mencetak hasil daripada prompt pertama dan tambah kod berikut di bawahnya:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # cetak respons
    print("Shopping list:")
    print(response.output_text)
    ```

    Perhatikan perkara berikut:

    - Kami membina prompt baru dengan menambah hasil daripada prompt pertama ke prompt baru:
    
        ```python

        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Kami membuat permintaan baru, tetapi juga mengambil kira bilangan token yang kami minta dalam prompt pertama, jadi kali ini kami katakan `max_output_tokens` ialah 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)

Sebaliknya semakin rendah nilai suhu, output lebih boleh diramal. Pertimbangkan sama ada anda mahu variasi pada output anda atau tidak.

   Untuk mengubah suhu, anda boleh gunakan parameter `temperature`. Contohnya, jika anda mahu gunakan suhu 0.5, anda boleh buat:

    response = client.responses.create(model=deployment, input=messages, store=False)
