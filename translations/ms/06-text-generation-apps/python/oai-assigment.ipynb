## Apakah itu aplikasi penjanaan teks?

### Aplikasi Konsol dan UI adalah terhad

## Jana teks

Cara untuk menjana teks adalah dengan menggunakan kaedah `responses.create`. Berikut contoh:

```python

print(response.output_text)
```

Dalam kod di atas, kami mencipta objek respons dan memasukkan model yang kami mahu gunakan dan prompt. Kemudian kami mencetak teks yang dijana.

### Respons sembang

Setakat ini, anda telah melihat bagaimana kami boleh menjana teks dari satu prompt. API Respons juga sangat sesuai untuk chatbot - anda menyediakan senarai mesej sebagai `input` untuk membina perbualan. Berikut contoh menggunakannya:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Perkataan tentang panjang token**. Kami harus pertimbangkan berapa banyak token yang diperlukan untuk menjana teks yang kami mahu. Token memerlukan kos, jadi bila boleh, kita perlu berjimat dengan jumlah token yang digunakan. Sebagai contoh, bolehkah kita merangka arahan supaya gunakan lebih sedikit token?

   Untuk mengubah token digunakan, anda boleh menggunakan parameter `max_output_tokens`. Sebagai contoh, jika anda mahu guna 100 token, anda boleh lakukan:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Eksperimen dengan temperature**. Temperature adalah sesuatu yang belum kami sebutkan sebelum ini tetapi adalah konteks penting untuk prestasi program kami. Semakin tinggi nilai temperature, semakin rawak output yang dijana. Sebaliknya, semakin rendah nilai temperature, semakin boleh diramalkan outputnya. Fikirkan sama ada anda mahu variasi dalam output atau tidak.

   Untuk menukar temperature, anda boleh gunakan parameter `temperature`. Sebagai contoh, jika anda mahu guna temperature 0.5, anda boleh lakukan:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Nota, semakin hampir ke 1.0, semakin pelbagai outputnya.
