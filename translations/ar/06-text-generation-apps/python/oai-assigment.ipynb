## ما هو تطبيق توليد النصوص؟

عادةً عند بناء تطبيق يكون له نوع من الواجهة مثل التالي:

- قائم على الأوامر. تطبيقات الكونسول هي تطبيقات نموذجية حيث تكتب أمرًا وينفذ مهمة. على سبيل المثال، `git` هو تطبيق قائم على الأوامر.
- واجهة المستخدم (UI). بعض التطبيقات تحتوي على واجهات مستخدم رسومية (GUIs) حيث تنقر الأزرار، تدخل نصًا، تختار خيارات والمزيد.

### تطبيقات الكونسول وواجهة المستخدم محدودة

قارنها بتطبيق قائم على الأوامر حيث تكتب أمرًا:

- **محدودة**. لا يمكنك كتابة أي أمر، فقط الأوامر التي يدعمها التطبيق.
- **محددة باللغة**. بعض التطبيقات تدعم العديد من اللغات، ولكن بشكل افتراضي يتم بناء التطبيق للغة محددة، حتى لو كان بإمكانك إضافة دعم لغات أخرى.

### فوائد تطبيقات توليد النصوص

فكيف يختلف تطبيق توليد النصوص؟

في تطبيق توليد النصوص، لديك مرونة أكبر، لست محدودًا بمجموعة أوامر أو لغة إدخال محددة. بدلاً من ذلك، يمكنك استخدام اللغة الطبيعية للتفاعل مع التطبيق. فائدة أخرى هي أنك تتفاعل مع مصدر بيانات تم تدريبه على كم هائل من المعلومات، بينما قد يكون التطبيق التقليدي محدودًا بما هو موجود في قاعدة بيانات.

### ماذا يمكنني أن أبني بتطبيق توليد النصوص؟

هناك أشياء كثيرة يمكنك بناؤها. على سبيل المثال:

- **دردشة ذكية**. دردشة ذكية تجيب على أسئلة حول مواضيع، مثل شركتك ومنتجاتها، يمكن أن تكون خيارًا جيدًا.
- **مساعد**. نماذج اللغة الكبيرة ممتازة في أشياء مثل تلخيص النصوص، استخراج المعلومات من النصوص، إنتاج نصوص مثل السير الذاتية والمزيد.
- **مساعد برمجي**. حسب نموذج اللغة الذي تستخدمه، يمكنك بناء مساعد برمجي يساعدك على كتابة الكود. على سبيل المثال، يمكنك استخدام منتجات مثل GitHub Copilot وكذلك ChatGPT لمساعدتك في كتابة الكود.

## توليد النص

طريقة توليد النص هي استخدام دالة `responses.create`. هذا مثال:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

في الكود أعلاه، ننشئ كائن استجابة ونمرر النموذج الذي نريد استخدامه والمطالبة. ثم نطبع النص المولد.

### ردود الدردشة

حتى الآن، رأيت كيف يمكننا توليد نص من مطالبة واحدة. API الاستجابات أيضًا مناسب جدًا للدردشات - حيث تزود قائمة من الرسائل كـ`input` لبناء محادثة. هذا مثال على استخدامه:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

المزيد عن هذه الوظيفة في فصل قادم.

الآن بعد أن قمنا بمحاكاة سيناريو معين، لنكتب كودًا يطابق السيناريو المعروض. للقيام بذلك، اتبع الخطوات التالية:

1. استخدم ملف *app.py* الموجود كنقطة بداية
1. حدد موقع المتغير `prompt` وغيّر كوده إلى التالي:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### التحسين بإضافة فلتر وقائمة تسوق

لدينا الآن تطبيق يعمل قادر على إنتاج وصفات ومرن لأنه يعتمد على مدخلات المستخدم، سواء على عدد الوصفات أو على المكونات المستخدمة.

لتحسينه أكثر، نريد إضافة ما يلي:

- **تصفية المكونات**. نريد أن نتمكن من تصفية المكونات التي لا نحبها أو التي نحن مصابون بحساسية تجاهها. لتحقيق هذا التغيير، يمكننا تعديل الطلب الحالي وإضافة شرط فلترة في نهايته هكذا:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```
