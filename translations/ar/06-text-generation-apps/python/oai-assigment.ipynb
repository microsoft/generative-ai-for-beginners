## ما هو تطبيق توليد النصوص؟

عادةً عند بناء تطبيق، يكون لديه نوع من الواجهة مثل الآتي:

- قائم على الأوامر. تطبيقات الكونسول هي تطبيقات نموذجية حيث تكتب أمرًا وينفذ مهمة. على سبيل المثال، `git` هو تطبيق قائم على الأوامر.
- واجهة المستخدم (UI). بعض التطبيقات لديها واجهات مستخدم رسومية (GUIs) حيث تنقر على أزرار، تدخل نصًا، تختار خيارات والمزيد.

### تطبيقات الكونسول وواجهة المستخدم محدودة

قارنها بتطبيق قائم على الأوامر حيث تكتب أمرًا:

- **إنها محدودة**. لا يمكنك كتابة أي أمر، فقط الأوامر التي يدعمها التطبيق.
- **خاص بلغة معينة**. بعض التطبيقات تدعم العديد من اللغات، لكن بشكل افتراضي يُبنى التطبيق للغة محددة، حتى إذا كان يمكنك إضافة دعم لغات أخرى.

### فوائد تطبيقات توليد النصوص

فكيف يكون تطبيق توليد النصوص مختلفًا؟

في تطبيق توليد النصوص، لديك مرونة أكبر، لست مقيدًا بمجموعة أوامر أو لغة إدخال معينة. بدلاً من ذلك يمكنك استخدام اللغة الطبيعية للتفاعل مع التطبيق. فائدة أخرى هي أنه بما أنك تتفاعل بالفعل مع مصدر بيانات تم تدريبه على مجموعة ضخمة من المعلومات، بينما قد يكون التطبيق التقليدي محدودًا بما هو موجود في قاعدة بيانات.

### ماذا يمكنني أن أبني باستخدام تطبيق توليد النصوص؟

هناك العديد من الأشياء التي يمكنك بناؤها. على سبيل المثال:

- **دردشة ذكية**. دردشة تجيب على الأسئلة حول مواضيع، مثل شركتك ومنتجاتها يمكن أن تكون خيارًا جيدًا.
- **مساعد**. نماذج اللغة الكبيرة ممتازة في أشياء مثل تلخيص النصوص، استخراج الأفكار من النص، إنتاج نصوص مثل السير الذاتية والمزيد.
- **مساعد برمجي**. اعتمادًا على نموذج اللغة الذي تستخدمه، يمكنك بناء مساعد برمجي يساعدك على كتابة الكود. على سبيل المثال، يمكنك استخدام منتج مثل GitHub Copilot بالإضافة إلى ChatGPT لمساعدتك في كتابة الكود.

## توليد النص

طريقة توليد النص هي استخدام دالة `responses.create`. إليك مثالًا:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

في الكود أعلاه، ننشئ كائن استجابة ونمرر النموذج الذي نريد استخدامه والمحفظ. ثم نطبع النص المولد.

### ردود المحادثة

حتى الآن، رأيت كيف يمكننا توليد نص من محفز واحد. واجهة برمجة التطبيقات Responses مناسبة أيضًا للدردشات — توفر قائمة رسائل كـ `input` لبناء محادثة. إليك مثالًا على استخدامها:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

المزيد عن هذه الوظيفة في فصل قادم.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
