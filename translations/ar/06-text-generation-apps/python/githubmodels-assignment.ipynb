# بناء تطبيقات توليد النصوص

لقد شاهدت حتى الآن من خلال هذا المنهج أن هناك مفاهيم أساسية مثل المطالبات وحتى تخصص كامل يُسمى "هندسة المطالبات". تدعم العديد من الأدوات التي يمكنك التفاعل معها مثل ChatGPT، Office 365، Microsoft Power Platform والمزيد، استخدام المطالبات لتحقيق شيء ما.

لكي تضيف مثل هذه التجربة إلى تطبيق، تحتاج إلى فهم مفاهيم مثل المطالبات، الاكتمالات واختيار مكتبة للعمل معها. هذا بالضبط ما ستتعلمه في هذا الفصل.

## مقدمة

في هذا الفصل، سوف:

- تتعلم عن مكتبة openai ومفاهيمها الأساسية.
- بناء تطبيق توليد نصوص باستخدام openai.
- فهم كيفية استخدام مفاهيم مثل المطالبة، درجة الحرارة، والرموز لبناء تطبيق توليد نصوص.

## أهداف التعلم

- شرح ما هو تطبيق توليد النصوص.

## ما هو تطبيق توليد النصوص؟

عادةً عندما تبني تطبيقًا، يكون له نوع من الواجهة مثل التالي:

- يعتمد على الأوامر. تطبيقات وحدة التحكم هي تطبيقات نموذجية حيث تكتب أمرًا ويتم تنفيذ مهمة. على سبيل المثال، `git` هو تطبيق يعتمد على الأوامر.
- واجهة المستخدم (UI). بعض التطبيقات لها واجهات مستخدم رسومية (GUIs) حيث تنقر على أزرار، تدخل نصًا، تختار خيارات والمزيد.

### تطبيقات وحدة التحكم وواجهة المستخدم محدودة

قارن ذلك بتطبيق يعتمد على الأوامر حيث تكتب أمرًا:

- **إنها محدودة**. لا يمكنك فقط كتابة أي أمر، فقط الأوامر التي يدعمها التطبيق.
- **محددة باللغة**. بعض التطبيقات تدعم عدة لغات، لكن بشكل افتراضي يتم بناء التطبيق للغة محددة، حتى لو يمكنك إضافة دعم للغات أخرى.

### فوائد تطبيقات توليد النصوص

فكيف يختلف تطبيق توليد النصوص؟

في تطبيق توليد النصوص، لديك مزيد من المرونة، لست مقيدًا بمجموعة أوامر معينة أو لغة إدخال محددة. بدلًا من ذلك، يمكنك استخدام اللغة الطبيعية للتفاعل مع التطبيق. فائدة أخرى هي لأنك بالفعل تتفاعل مع مصدر بيانات تم تدريبه على مجموعة ضخمة من المعلومات، بينما قد يكون تطبيق تقليدي محدودًا فيما هو موجود في قاعدة بيانات.

### ماذا يمكنني أن أبني بتطبيق توليد النصوص؟

هناك العديد من الأشياء التي يمكنك بناؤها. على سبيل المثال:

- **دردشة آلية**. دردشة تجيب عن أسئلة حول مواضيع، مثل شركتك ومنتجاتها قد تكون مناسبة جيدة.
- **مساعد**. النماذج اللغوية الكبيرة ممتازة في أشياء مثل تلخيص النصوص، استخراج الأفكار من النصوص، إنتاج نصوص مثل السير الذاتية والمزيد.
- **مساعد ترميز**. اعتمادًا على نموذج اللغة الذي تستخدمه، يمكنك بناء مساعد ترميز يساعدك في كتابة الشيفرة. على سبيل المثال، يمكنك استخدام منتج مثل GitHub Copilot وكذلك ChatGPT لمساعدتك في كتابة الشيفرة.

بعد أن قمنا بتجسيد سيناريو، دعنا نكتب الكود ليطابق السيناريو المعروض. للقيام بذلك، اتبع الخطوات التالية:

1. أنشئ متغيرًا باسم `prompt` وغير الكود النموذجي كما يلي:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
