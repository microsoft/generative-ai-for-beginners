# بناء تطبيقات توليد النصوص

لقد رأيت حتى الآن من خلال هذه المناهج أن هناك مفاهيم أساسية مثل المحفزات وحتى تخصص كامل يُسمى "هندسة المحفزات". العديد من الأدوات التي يمكنك التفاعل معها مثل ChatGPT، Office 365، Microsoft Power Platform والمزيد، تدعمك باستخدام المحفزات لتحقيق شيء ما.

لكي تضيف مثل هذه التجربة إلى تطبيق، تحتاج إلى فهم مفاهيم مثل المحفزات، الإكمالات واختيار مكتبة للعمل معها. هذا بالضبط ما ستتعلمه في هذا الفصل.

## المقدمة

في هذا الفصل، ستتعلم:

- التعرف على مكتبة openai ومفاهيمها الأساسية.
- بناء تطبيق لتوليد النصوص باستخدام openai.
- فهم كيفية استخدام مفاهيم مثل المحفز، درجة الحرارة، والرموز لبناء تطبيق توليد نصوص.

## أهداف التعلم

في نهاية هذا الدرس، ستكون قادرًا على:

- شرح ماهية تطبيق توليد النصوص.
- بناء تطبيق توليد نصوص باستخدام openai.
- ضبط تطبيقك لاستخدام المزيد أو القليل من الرموز بالإضافة إلى تغيير درجة الحرارة، للحصول على مخرجات متنوعة.

## ما هو تطبيق توليد النصوص؟

عادةً عند بناء تطبيق، يكون لديه نوع من الواجهة مثل التالية:

- معتمد على الأوامر. التطبيقات النصية هي تطبيقات عادية حيث تكتب أمرًا ويتم تنفيذ مهمة. على سبيل المثال، `git` هو تطبيق معتمد على الأوامر.
- واجهة المستخدم (UI). بعض التطبيقات لديها واجهات مستخدم رسومية (GUIs) حيث تضغط أزرارًا، تدخل نصًا، تختار خيارات وأكثر.

### التطبيقات النصية وواجهات المستخدم محدودة

قارن ذلك مع تطبيق معتمد على الأوامر حيث تكتب أمرًا:

- **إنه محدود**. لا يمكنك فقط كتابة أي أمر، فقط الأوامر التي يدعمها التطبيق.
- **محدد اللغة**. بعض التطبيقات تدعم عدة لغات، لكن بشكل افتراضي التطبيق مبني للغة محددة، حتى لو كان يمكنك إضافة دعم لغات أخرى.

### فوائد تطبيقات توليد النصوص

فكيف يختلف تطبيق توليد النصوص؟

في تطبيق توليد النصوص، لديك مزيد من المرونة، أنت لست مقيدًا بمجموعة من الأوامر أو لغة إدخال محددة. بدلاً من ذلك، يمكنك استخدام اللغة الطبيعية للتفاعل مع التطبيق. ميزة أخرى هي أنك تتفاعل مع مصدر بيانات تم تدريبه على مجموعة هائلة من المعلومات، بينما التطبيقات التقليدية قد تكون محدودة بما يتوفر في قاعدة بيانات.

### ماذا يمكنني أن أبني باستخدام تطبيق توليد النصوص؟

هناك أشياء كثيرة يمكنك بناؤها. على سبيل المثال:

- **دردشة آلية**. دردشة تجيب عن أسئلة حول مواضيع مثل شركتك ومنتجاتها قد تكون مناسبة جدًا.
- **مساعد**. نماذج اللغة الكبيرة رائعة في أشياء مثل تلخيص النصوص، استخراج الأفكار من النصوص، إنتاج نصوص مثل السير الذاتية وأكثر.
- **مساعد كود**. اعتمادًا على نموذج اللغة الذي تستخدمه، يمكنك بناء مساعد كود يساعدك في كتابة الأكواد. على سبيل المثال، يمكنك استخدام منتج مثل GitHub Copilot وأيضًا ChatGPT لمساعدتك في كتابة الكود.

## تمرين - بناء مولد وصفات

الآن بعد أن قمنا بمحاكاة السيناريو، دعنا نكتب الكود ليتوافق مع السيناريو الموضح. للقيام بذلك، اتبع الخطوات التالية:

1. استخدم الملف الحالي كنقطة بداية
1. أنشئ متغيرًا باسم `prompt` وقم بتغيير الكود النموذجي كما يلي:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
