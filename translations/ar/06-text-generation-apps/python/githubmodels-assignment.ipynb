# بناء تطبيقات توليد النصوص

لقد رأيت حتى الآن من خلال هذا المنهج أن هناك مفاهيم أساسية مثل المحفزات (prompts) وحتى مجال كامل يُسمى "هندسة المحفزات". العديد من الأدوات التي يمكنك التفاعل معها مثل ChatGPT، Office 365، Microsoft Power Platform وغيرها، تدعم استخدامك للمحفزات لتحقيق هدف معين.

لكي تضيف مثل هذه التجربة إلى تطبيقك، تحتاج إلى فهم مفاهيم مثل المحفزات، الإكمالات (completions) واختيار مكتبة للعمل معها. هذا بالضبط ما ستتعلمه في هذا الفصل.

## المقدمة

في هذا الفصل، سوف:

- تتعرف على مكتبة openai ومفاهيمها الأساسية.
- تبني تطبيق توليد نصوص باستخدام openai.
- تفهم كيفية استخدام مفاهيم مثل prompt، ودرجة الحرارة (temperature)، والتوكنات (tokens) لبناء تطبيق توليد نصوص.

## أهداف التعلم

في نهاية هذا الدرس، ستكون قادرًا على:

- شرح ما هو تطبيق توليد النصوص.
- بناء تطبيق توليد نصوص باستخدام openai.
- ضبط تطبيقك لاستخدام عدد أكبر أو أقل من التوكنات وأيضًا تغيير درجة الحرارة للحصول على نتائج متنوعة.

## ما هو تطبيق توليد النصوص؟

عادةً عندما تبني تطبيقًا يكون له نوع من الواجهة مثل التالي:

- قائم على الأوامر. تطبيقات الكونسول هي تطبيقات نموذجية حيث تكتب أمرًا ويقوم التطبيق بتنفيذ مهمة. على سبيل المثال، `git` هو تطبيق قائم على الأوامر.
- واجهة المستخدم (UI). بعض التطبيقات لديها واجهات رسومية حيث يمكنك النقر على الأزرار، إدخال نص، اختيار خيارات والمزيد.

### تطبيقات الكونسول وواجهة المستخدم محدودة

قارن ذلك بتطبيق قائم على الأوامر حيث تكتب أمرًا:

- **محدود**. لا يمكنك كتابة أي أمر، فقط الأوامر التي يدعمها التطبيق.
- **مخصص للغة معينة**. بعض التطبيقات تدعم عدة لغات، لكن بشكل افتراضي التطبيق مبني للغة معينة، حتى لو كان بإمكانك إضافة دعم للغات أخرى.

### فوائد تطبيقات توليد النصوص

فكيف يختلف تطبيق توليد النصوص؟

في تطبيق توليد النصوص، لديك مرونة أكبر، لست مقيدًا بمجموعة أوامر أو لغة إدخال محددة. بدلاً من ذلك، يمكنك استخدام اللغة الطبيعية للتفاعل مع التطبيق. فائدة أخرى هي أنك تتعامل مع مصدر بيانات تم تدريبه على كم هائل من المعلومات، بينما التطبيق التقليدي قد يكون محدودًا بما هو موجود في قاعدة البيانات فقط.

### ماذا يمكنني أن أبني باستخدام تطبيق توليد النصوص؟

هناك العديد من الأشياء التي يمكنك بناؤها. على سبيل المثال:

- **شات بوت**. شات بوت يجيب على الأسئلة حول مواضيع مثل شركتك ومنتجاتها يمكن أن يكون خيارًا جيدًا.
- **مساعد**. النماذج اللغوية الكبيرة ممتازة في تلخيص النصوص، استخراج الأفكار من النصوص، إنتاج نصوص مثل السير الذاتية والمزيد.
- **مساعد برمجي**. حسب النموذج اللغوي الذي تستخدمه، يمكنك بناء مساعد برمجي يساعدك في كتابة الكود. على سبيل المثال، يمكنك استخدام منتج مثل GitHub Copilot أو ChatGPT لمساعدتك في كتابة الكود.

## تمرين - بناء مولّد وصفات

بعد أن استعرضنا سيناريو معين، دعونا نكتب الشيفرة التي تتوافق مع هذا السيناريو. للقيام بذلك، اتبع الخطوات التالية:

1. استخدم الملف الحالي كنقطة بداية
1. أنشئ متغير باسم `prompt` وقم بتعديل الشيفرة التجريبية كما يلي:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

## اختبار المعرفة

1. يتحكم في مدى عشوائية الناتج.
1. يتحكم في حجم الاستجابة.
1. يتحكم في عدد التوكنات المستخدمة.

الإجابة: 1

ما هي أفضل طريقة لتخزين الأسرار مثل مفاتيح API؟
