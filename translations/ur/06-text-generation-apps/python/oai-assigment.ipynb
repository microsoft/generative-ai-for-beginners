## متن تخلیق کریں

متن تخلیق کرنے کا طریقہ `responses.create` طریقہ استعمال کرنا ہے۔ یہاں ایک مثال ہے:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

اوپر والے کوڈ میں، ہم ایک response آبجیکٹ بناتے ہیں اور ماڈل اور پرامپٹ دیتے ہیں۔ پھر تخلیق شدہ متن کو پرنٹ کرتے ہیں۔

### چیٹ جوابات

اب تک، آپ نے دیکھا کہ ہم ایک واحد پرامپٹ سے متن کیسے تخلیق کر سکتے ہیں۔ Responses API چیٹ بوٹس کے لیے بھی بہت موزوں ہے - آپ `input` کے طور پر پیغامات کی فہرست فراہم کرتے ہیں تاکہ بات چیت بن سکے۔ یہاں ایک مثال ہے:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

اس خصوصیت پر مزید اگلے باب میں بات ہوگی۔

## مشق - آپ کی پہلی متن تخلیق کرنے والی ایپ

اب جب کہ ہم نے OpenAI سروس کی ترتیب اور کنفیگریشن سیکھ لی ہے، وقت ہے کہ آپ اپنی پہلی متن تخلیق کرنے والی ایپ بنائیں۔ اپنی ایپ بنانے کے لیے، ان مراحل پر عمل کریں:


1. ایک ورچوئل ماحول بنائیں اور openai انسٹال کریں:

  > [!NOTE] اگر آپ یہ نوٹ بک Codespaces پر یا Devcontainer کے اندر چلاتے ہیں تو یہ قدم ضروری نہیں ہے


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> اگر آپ ونڈوز استعمال کر رہے ہیں تو `source venv/bin/activate` کی جگہ `venv\Scripts\activate` ٹائپ کریں۔

> [!NOTE]
> اپنا OpenAI کلید تلاش کرنے کے لیے https://platform.openai.com/settings/organization/api-keys پر جائیں اور `API keys` تلاش کریں۔ آپ وہاں نئی کلید بنا سکتے ہیں اور فوراً اس کی قیمت کاپی کر سکتے ہیں۔


1. ایک *app.py* فائل بنائیں اور اسے درج ذیل کوڈ دیں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    آپ کو مندرجہ ذیل جیسا آؤٹ پٹ دیکھنا چاہیے:

    ```output
     بہت زیادہ اداس _____.

    ایک دفعہ کا ذکر ہے کہ ایک بہت اداس میری میڈ تھی۔
    ```


## مختلف چیزوں کے لیے مختلف قسم کے پرامپٹس  

اب آپ نے دیکھا کہ پرامپٹ استعمال کرتے ہوئے متن کیسے تیار کیا جاتا ہے۔ آپ کا ایک پروگرام بھی چل رہا ہے جسے آپ مختلف اقسام کے متن تیار کرنے کے لیے ترمیم اور تبدیل کر سکتے ہیں۔

پرامپٹس مختلف قسم کے کاموں کے لیے استعمال کیے جا سکتے ہیں۔ مثال کے طور پر:  

- **کسی قسم کا متن تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئی کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔  
- **معلومات تلاش کرنا**۔ آپ پرامپٹس کو اس طرح کی معلومات تلاش کرنے کے لیے استعمال کر سکتے ہیں جیسے کہ 'ویب ڈویلپمنٹ میں CORS کا کیا مطلب ہے؟'  
- **کوڈ تیار کرنا**۔ آپ پرامپٹس کو کوڈ تیار کرنے کے لیے استعمال کر سکتے ہیں، مثلاً ای میلز کی جانچ کے لیے ریگولر ایکسپریشن تیار کرنا یا پورا پروگرام تیار کرنا، جیسے ویب ایپلیکیشن؟

## مشق - ریسیپی جنریٹر بنائیں

اب جب کہ ہم نے ایک منظرنامہ ادا کر لیا ہے، آئیے کوڈ لکھیں جو دکھائے گئے منظرنامے سے میل کھاتا ہو۔ ایسا کرنے کے لیے، درج ذیل مراحل پر عمل کریں:

1. موجودہ *app.py* فائل کو نقطہ آغاز کے طور پر استعمال کریں
1. `prompt` متغیر کو تلاش کریں اور اس کا کوڈ درج ذیل میں تبدیل کریں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> نوٹ، آپ کا LLM غیر متعین ہے، اس لیے آپ کو ہر بار پروگرام چلانے پر مختلف نتائج مل سکتے ہیں۔

اچھا، آئیے دیکھتے ہیں کہ ہم چیزوں کو کیسے بہتر بنا سکتے ہیں۔ چیزوں کو بہتر بنانے کے لیے، ہم چاہتے ہیں کہ کوڈ لچکدار ہو، تاکہ اجزاء اور ترکیبوں کی تعداد میں بہتری اور تبدیلی کی جا سکے۔


1. آئیں کوڈ کو مندرجہ ذیل طریقے سے تبدیل کریں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### فلٹر اور خریداری کی فہرست شامل کر کے بہتری لائیں

اسے مزید بہتر بنانے کے لیے، ہم درج ذیل شامل کرنا چاہتے ہیں:

    old_prompt_result = response.output_text

    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # جواب چھاپیں
    print("Shopping list:")
    print(response.output_text)

        new_prompt = f"{old_prompt_result} {prompt}"

- **ٹوکن کی تعداد کے بارے میں ایک بات**۔ ہمیں غور کرنا چاہیے کہ ہمیں وہ متن پیدا کرنے کے لیے کتنے ٹوکنز کی ضرورت ہے جو ہم چاہتے ہیں۔ ٹوکنز کی قیمت ہوتی ہے، اس لیے جہاں ممکن ہو، ہمیں ٹوکنز کی مقدار کو اقتصادی بنانا چاہیے۔ مثال کے طور پر، کیا ہم پرامپٹ کو اس طرح سیٹ کر سکتے ہیں کہ کم ٹوکنز استعمال ہوں؟
