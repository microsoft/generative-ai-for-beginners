## ٹیکسٹ جنریشن ایپ کیا ہے؟

عام طور پر جب آپ کوئی ایپ بناتے ہیں تو اس کی کوئی نہ کوئی انٹرفیس ہوتی ہے جیسے:

- کمانڈ بیسڈ۔ کنسول ایپس عام طور پر وہ ایپس ہوتی ہیں جن میں آپ کمانڈ لکھتے ہیں اور وہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ بیسڈ ایپ ہے۔
- یوزر انٹرفیس (UI)۔ کچھ ایپس میں گرافیکل یوزر انٹرفیس (GUI) ہوتا ہے جہاں آپ بٹن کلک کرتے ہیں، ٹیکسٹ لکھتے ہیں، آپشنز منتخب کرتے ہیں وغیرہ۔

### کنسول اور UI ایپس کی حدود

اس کا موازنہ ایک کمانڈ بیسڈ ایپ سے کریں جس میں آپ کمانڈ لکھتے ہیں:

- **اس کی حدود ہیں**۔ آپ کوئی بھی کمانڈ نہیں لکھ سکتے، صرف وہی کمانڈز جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس کئی زبانیں سپورٹ کرتی ہیں، لیکن عام طور پر ایپ ایک خاص زبان کے لیے بنائی جاتی ہے، چاہے آپ مزید زبانیں شامل کر بھی لیں۔

### ٹیکسٹ جنریشن ایپس کے فائدے

تو ٹیکسٹ جنریشن ایپ کس طرح مختلف ہے؟

ٹیکسٹ جنریشن ایپ میں آپ کو زیادہ آزادی ملتی ہے، آپ کسی خاص کمانڈ یا زبان تک محدود نہیں ہوتے۔ اس کے بجائے، آپ قدرتی زبان میں ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ آپ ایک ایسے ڈیٹا سورس سے بات کر رہے ہوتے ہیں جو پہلے ہی بہت زیادہ معلومات پر تربیت یافتہ ہے، جبکہ روایتی ایپ صرف ڈیٹا بیس تک محدود ہو سکتی ہے۔

## ٹیکسٹ جنریٹ کریں

ٹیکسٹ جنریٹ کرنے کا طریقہ یہ ہے کہ `chat.completion` کلاس استعمال کریں۔ یہاں ایک مثال ہے:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

اوپر والے کوڈ میں ہم ایک کمپلیشن آبجیکٹ بناتے ہیں اور اس میں ماڈل اور پرامپٹ دیتے ہیں۔ پھر ہم جنریٹڈ ٹیکسٹ پرنٹ کرتے ہیں۔

### چیٹ کمپلیشنز

اب تک آپ نے دیکھا کہ ہم `Completion` کا استعمال کر کے ٹیکسٹ جنریٹ کر رہے تھے۔ لیکن ایک اور کلاس ہے `ChatCompletion` جو چیٹ بوٹس کے لیے زیادہ موزوں ہے۔ اس کا استعمال کرنے کی ایک مثال یہ ہے:


## مشق - اپنی پہلی ٹیکسٹ جنریشن ایپ

1. ایک ورچوئل ماحول بنائیں اور openai انسٹال کریں:

  > [!NOTE] اگر آپ یہ نوٹ بک Codespaces یا Devcontainer میں چلا رہے ہیں تو یہ مرحلہ ضروری نہیں۔

> [!NOTE]
> اگر آپ ونڈوز استعمال کر رہے ہیں تو `venv\Scripts\activate` لکھیں بجائے اس کے کہ `source venv/bin/activate`۔

> [!NOTE]
> اپنا OpenAI کلید تلاش کرنے کے لیے https://platform.openai.com/settings/organization/api-keys پر جائیں اور `API keys` تلاش کریں۔ آپ وہاں نیا کلید بنا سکتے ہیں اور فوراً اس کی ویلیو کاپی کر سکتے ہیں۔


1. ایک *app.py* فائل بنائیں اور اس میں درج ذیل کوڈ لکھیں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## مختلف قسم کے پرامپٹس، مختلف کاموں کے لیے

اب آپ نے دیکھ لیا ہے کہ کس طرح ایک پرامپٹ کے ذریعے ٹیکسٹ تیار کیا جا سکتا ہے۔ آپ کے پاس ایک پروگرام بھی ہے جسے آپ اپنی مرضی کے مطابق بدل سکتے ہیں اور مختلف قسم کا ٹیکسٹ تیار کر سکتے ہیں۔

پرامپٹس کو مختلف قسم کے کاموں کے لیے استعمال کیا جا سکتا ہے۔ مثلاً:

- **کسی خاص قسم کا ٹیکسٹ تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔
- **معلومات تلاش کرنا**۔ آپ پرامپٹس کے ذریعے معلومات تلاش کر سکتے ہیں، جیسے یہ مثال: 'ویب ڈیولپمنٹ میں CORS کا کیا مطلب ہے؟'
- **کوڈ تیار کرنا**۔ آپ پرامپٹس کے ذریعے کوڈ تیار کر سکتے ہیں، مثلاً ای میل کی تصدیق کے لیے ریگولر ایکسپریشن بنانا یا پھر پورا پروگرام تیار کرنا، جیسے ویب ایپ۔

## مشق - ایک نسخہ جنریٹر بنائیں

اب جب کہ ہم نے ایک منظرنامہ آزما لیا ہے، آئیے اس منظرنامے کے مطابق کوڈ لکھتے ہیں۔ اس کے لیے یہ اقدامات کریں:

1. موجودہ *app.py* فائل کو نقطہ آغاز کے طور پر استعمال کریں
1. `prompt` متغیر کو تلاش کریں اور اس کا کوڈ درج ذیل میں تبدیل کریں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

زبردست، اب دیکھتے ہیں کہ ہم چیزوں کو کیسے بہتر بنا سکتے ہیں۔ بہتری کے لیے، ہم چاہتے ہیں کہ کوڈ لچکدار ہو، تاکہ اجزاء اور ترکیبوں کی تعداد کو آسانی سے بڑھایا اور تبدیل کیا جا سکے۔


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
