اگر آپ کسی ایپ میں ایسا تجربہ شامل کرنا چاہتے ہیں، تو آپ کو پرامپٹس، کمپلیشنز جیسے تصورات کو سمجھنا ہوگا اور استعمال کرنے کے لیے کوئی لائبریری منتخب کرنی ہوگی۔ بالکل یہی آپ اس باب میں سیکھیں گے۔

## ٹیکسٹ جنریشن ایپ کیا ہے؟

عام طور پر جب آپ کوئی ایپ بناتے ہیں تو اس میں کچھ قسم کا انٹرفیس ہوتا ہے جیسا کہ درج ذیل:

- کمانڈ بیسڈ۔ کنسول ایپس عام ایپس ہوتی ہیں جہاں آپ ایک کمانڈ ٹائپ کرتے ہیں اور یہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ بیسڈ ایپ ہے۔
- یوزر انٹرفیس (UI)۔ بعض ایپس گرافیکل یوزر انٹرفیسز (GUIs) رکھتی ہیں جہاں آپ بٹن کلک کرتے ہیں، متن داخل کرتے ہیں، آپشن منتخب کرتے ہیں اور بھی بہت کچھ۔

### کنسول اور UI ایپس محدود ہیں

اس کا موازنہ ایک کمانڈ بیسڈ ایپ سے کریں جہاں آپ کمانڈ ٹائپ کرتے ہیں:

- **یہ محدود ہے**۔ آپ کوئی بھی کمانڈ ٹائپ نہیں کر سکتے، صرف وہی جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس متعدد زبانیں سپورٹ کرتی ہیں، لیکن ڈیفالٹ طور پر ایپ کسی مخصوص زبان کے لیے بنائی گئی ہوتی ہے، چاہے آپ مزید زبانوں کی حمایت شامل کر سکیں۔

### ٹیکسٹ جنریشن ایپس کے فوائد

تو ٹیکسٹ جنریشن ایپ کیسے مختلف ہوتی ہے؟

ایک ٹیکسٹ جنریشن ایپ میں آپ زیادہ لچک رکھتے ہیں، آپ کسی مخصوص کمانڈز یا ان پٹ زبان تک محدود نہیں ہوتے۔ آپ قدرتی زبان استعمال کرکے ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ آپ ایک ایسے ڈیٹا سورس سے بات چیت کر رہے ہیں جو بہت زیادہ معلومات کے ذخیرے پر تربیت یافتہ ہے، جب کہ روایتی ایپ ڈیٹا بیس میں محدود ہو سکتی ہے۔

### میں ٹیکسٹ جنریشن ایپ سے کیا بنا سکتا ہوں؟

آپ بہت سی چیزیں بنا سکتے ہیں۔ جیسے:

- **ایک چیٹ بوٹ**۔ ایک چیٹ بوٹ جو آپ کی کمپنی اور اس کی مصنوعات کے بارے میں سوالات کے جواب دیتا ہو، ایک اچھا انتخاب ہو سکتا ہے۔
- **مددگار**۔ LLMs ایسے کاموں میں بہترین ہیں جیسے متن کا خلاصہ کرنا، متن سے بصیرتیں حاصل کرنا، ریزیومے جیسے متن تیار کرنا اور دیگر۔
- **کوڈ اسسٹنٹ**۔ آپ جس زبان کا ماڈل استعمال کرتے ہیں اس کے حساب سے آپ کوڈ لکھنے میں مدد دینے والا کوڈ اسسٹنٹ بنا سکتے ہیں۔ مثال کے طور پر، آپ GitHub Copilot اور ChatGPT جیسی مصنوعات استعمال کر کے کوڈ لکھنے میں مدد حاصل کر سکتے ہیں۔

## میں کیسے شروع کر سکتا ہوں؟

- لائبریری استعمال کریں۔ لائبریریاں API کالز کو encapsulate کرتی ہیں اور استعمال کو آسان بناتی ہیں۔

### OpenAI ترتیب دیں

## ٹیکسٹ تیار کریں

ٹیکسٹ تیار کرنے کا طریقہ `responses.create` میتھڈ استعمال کرنا ہے۔ مثال یہاں ہے:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

اوپر دیے گئے کوڈ میں، ہم ایک response object بناتے ہیں اور اس میں ماڈل اور پرامپٹ دے کر تیار کردہ ٹیکسٹ پرنٹ کرتے ہیں۔

### چیٹ جوابات

اب تک، آپ نے دیکھا کہ ہم ایک پرامپٹ سے ٹیکسٹ کیسے بنا سکتے ہیں۔ Responses API چیٹ بوٹس کے لیے بھی بہت موزوں ہے - آپ میسجز کی ایک فہرست `input` کے طور پر دیتے ہیں تاکہ بات چیت بنائی جا سکے۔ مثال یہاں ہے:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

اس فعالیت پر مزید تفصیل آنے والے باب میں دی جائے گی۔

## مشق - آپ کی پہلی ٹیکسٹ جنریشن ایپ

اب جب کہ ہم نے OpenAI سروس سیٹ اپ اور ترتیب دینا سیکھ لیا ہے، وقت ہے کہ آپ اپنی پہلی ٹیکسٹ جنریشن ایپ بنائیں۔ اپنی ایپ بنانے کے لیے درج ذیل مراحل کی پیروی کریں:


1. ایک ورچوئل ماحول بنائیں اور openai انسٹال کریں:

  > [!NOTE] اگر آپ یہ نوٹ بک Codespaces پر یا Devcontainer میں چلاتے ہیں تو یہ مرحلہ ضروری نہیں ہے


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> اگر آپ ونڈوز استعمال کر رہے ہیں تو `source venv/bin/activate` کی جگہ `venv\Scripts\activate` ٹائپ کریں۔

> [!NOTE]
> اپنا OpenAI کلید تلاش کرنے کے لیے https://platform.openai.com/settings/organization/api-keys پر جائیں اور `API keys` تلاش کریں۔ آپ وہاں نیا کلید بنا سکتے ہیں اور فوراً اس کی ویلیو کاپی کر سکتے ہیں۔


1. ایک *app.py* فائل بنائیں اور اسے مندرجہ ذیل کوڈ دیں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    آپ کو مندرجہ ذیل جیسا آؤٹ پٹ نظر آنا چاہئے:

    ```output
     بہت ناخوش _____.

    ایک دفعہ کا ذکر ہے کہ ایک بہت ناخوش مرمیڈ تھی۔
    ```


## مختلف کاموں کے لیے مختلف قسم کے پرومپٹس

اب آپ نے دیکھا کہ پرومپٹ کے ذریعے متن کیسے تیار کیا جاتا ہے۔ آپ کے پاس ایک پروگرام بھی چل رہا ہے جسے آپ مختلف قسم کے متن تیار کرنے کے لیے ترمیم اور تبدیل کر سکتے ہیں۔

پرومپٹس مختلف طرح کے کاموں کے لیے استعمال کیے جا سکتے ہیں۔ مثلاً:

- **کسی قسم کا متن تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔
- **معلومات تلاش کرنا**۔ آپ پرومپٹس کا استعمال معلومات تلاش کرنے کے لیے کر سکتے ہیں جیسے کہ اس مثال میں 'ویب ڈویلپمنٹ میں CORS کا کیا مطلب ہے؟'
- **کوڈ تیار کرنا**۔ آپ پرومپٹس کا استعمال کوڈ تیار کرنے کے لیے کر سکتے ہیں، مثلاً ای میلز کی تصدیق کے لیے ریگولر ایکسپریشن بنانا یا پورا پروگرام تیار کرنا جیسے کہ ایک ویب ایپ؟

## مشق - ایک ترکیب جنریٹر بنائیں

اب جب کہ ہم نے ایک منظرنامہ کھیل لیا ہے، آئیے اس منظرنامے سے میل کھانے والا کوڈ لکھیں۔ ایسا کرنے کے لیے، درج ذیل مراحل پر عمل کریں:

1. موجودہ *app.py* فائل کو ایک نقطہ آغاز کے طور پر استعمال کریں
1. `prompt` متغیر کو تلاش کریں اور اس کا کوڈ درج ذیل میں تبدیل کریں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

بہت خوب، آئیے دیکھتے ہیں کہ ہم چیزوں کو کیسے بہتر بنا سکتے ہیں۔ چیزوں کو بہتر بنانے کے لیے، ہم چاہتے ہیں کہ کوڈ لچکدار ہو، تاکہ اجزاء اور ترکیبوں کی تعداد کو بہتر اور تبدیل کیا جا سکے۔
