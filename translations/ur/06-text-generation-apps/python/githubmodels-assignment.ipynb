# ٹیکسٹ جنریشن ایپس بنائیں

آپ نے اب تک اس نصاب کے ذریعے دیکھا ہے کہ بنیادی تصورات جیسے پرامپٹس اور یہاں تک کہ ایک مکمل شعبہ "پرومپٹ انجینئرنگ" بھی ہے۔ بہت سے ٹولز جن کے ساتھ آپ بات چیت کر سکتے ہیں جیسے ChatGPT، Office 365، Microsoft Power Platform اور مزید، آپ کی مدد کرتے ہیں کہ آپ کچھ حاصل کرنے کے لئے پرامپٹس استعمال کریں۔

آپ کو ایسی ایپ میں ایسا تجربہ شامل کرنے کے لیے، آپ کو پرامپٹس، کمپلیشنز جیسے تصورات کو سمجھنا ہوگا اور کام کرنے کے لیے کوئی لائبریری منتخب کرنا ہوگی۔ بالکل یہی آپ اس باب میں سیکھیں گے۔

## تعارف

اس باب میں، آپ کریں گے:

- openai لائبریری اور اس کے بنیادی تصورات کے بارے میں سیکھیں۔
- openai استعمال کرتے ہوئے ایک ٹیکسٹ جنریشن ایپ بنائیں۔
- سمجھیں کہ ٹیکسٹ جنریشن ایپ بنانے کے لیے پرامپٹ، ٹیمپریچر، اور ٹوکنز جیسے تصورات کو کیسے استعمال کیا جاتا ہے۔

## تعلیمی اہداف

اس سبق کے آخر میں، آپ قابل ہوں گے:

- وضاحت کریں کہ ٹیکسٹ جنریشن ایپ کیا ہے۔
- openai استعمال کرتے ہوئے ایک ٹیکسٹ جنریشن ایپ بنائیں۔
- اپنی ایپ کو زیادہ یا کم ٹوکنز استعمال کرنے کے لیے کنفیگر کریں اور ٹیمپریچر بھی بدلیں، تاکہ منوع آؤٹ پٹ حاصل ہو۔

## ٹیکسٹ جنریشن ایپ کیا ہے؟

عام طور پر جب آپ کوئی ایپ بناتے ہیں تو اس کا کوئی انٹرفیس ہوتا ہے جیسے ذیل میں:

- کمانڈ بیسڈ۔ کنسول ایپس عام ایپس ہیں جہاں آپ کمانڈ ٹائپ کرتے ہیں اور یہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ بیسڈ ایپ ہے۔
- یوزر انٹرفیس (UI). کچھ ایپس کے گرافیکل یوزر انٹرفیس (GUIs) ہوتے ہیں جہاں آپ بٹن کلک کرتے ہیں، متن داخل کرتے ہیں، اختیارات منتخب کرتے ہیں وغیرہ۔

### کنسول اور UI ایپس محدود ہوتی ہیں

اس کا موازنہ ایک کمانڈ بیسڈ ایپ سے کریں جہاں آپ کمانڈ لکھتے ہیں:

- **یہ محدود ہے**۔ آپ کسی بھی کمانڈ کو نہیں لکھ سکتے، صرف وہ جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس بہت سی زبانیں سپورٹ کرتی ہیں، لیکن ڈیفالٹ کے طور پر ایپ کسی مخصوص زبان کے لیے بنائی گئی ہوتی ہے، چاہے آپ مزید زبانوں کی سپورٹ شامل کر سکیں۔

### ٹیکسٹ جنریشن ایپس کے فوائد

تو ٹیکسٹ جنریشن ایپ کیسے مختلف ہے؟

ٹیکسٹ جنریشن ایپ میں آپ کے پاس زیادہ لچکدار ہوتا ہے، آپ کمانڈز کی محدود تعداد یا مخصوص ان پٹ زبان تک محدود نہیں ہوتے۔ اس کے بجائے، آپ قدرتی زبان استعمال کرکے ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ چونکہ آپ پہلے ہی ایسی ڈیٹا سورس کے ساتھ تعامل کر رہے ہیں جو وسیع معلومات کے ذخیرے پر تربیت یافتہ ہے، جبکہ روایتی ایپ ڈیٹا بیس میں محدود ہو سکتی ہے۔

### ٹیکسٹ جنریشن ایپ سے میں کیا بنا سکتا ہوں؟

آپ بہت سی چیزیں بنا سکتے ہیں۔ مثال کے طور پر:

- **ایک چیٹ بوٹ**۔ ایک چیٹ بوٹ جو موضوعات کے بارے میں سوالات کے جواب دیتا ہے، جیسے آپ کی کمپنی اور اس کی مصنوعات کے بارے میں، یہ ایک اچھا میل ہو سکتا ہے۔
- **مددگار**۔ LLMs متن کا خلاصہ بنانے، متن سے بصیرت حاصل کرنے، ریزیومے اور مزید جیسے متن تیار کرنے میں بہت اچھے ہیں۔
- **کوڈ معاون**۔ جس زبان ماڈل کا آپ استعمال کرتے ہیں اس پر منحصر ہے، آپ ایک کوڈ معاون بنا سکتے ہیں جو کوڈ لکھنے میں مدد کرتا ہے۔ مثال کے طور پر، آپ GitHub Copilot اور ChatGPT جیسے پروڈکٹس استعمال کر سکتے ہیں جو آپ کو کوڈ لکھنے میں مدد دیتے ہیں۔

## میں کیسے شروع کر سکتا ہوں؟

خیر، آپ کو LLM کے ساتھ انٹیگریٹ کرنے کا طریقہ تلاش کرنا ہوگا جو عام طور پر درج ذیل دو طریقہ کار پر مشتمل ہوتا ہے:

- API استعمال کریں۔ یہاں آپ اپنے پرامپٹ کے ساتھ ویب ریکویسٹ بناتے ہیں اور پیداکردہ متن واپس حاصل کرتے ہیں۔
- لائبریری استعمال کریں۔ لائبریریاں API کالز کو لپیٹتی ہیں اور انہیں استعمال کرنا آسان بناتی ہیں۔

## لائبریریاں/SDKs

LLMs کے ساتھ کام کرنے کے لیے کچھ مشہور لائبریریاں ہیں جیسے:

- **openai**، یہ لائبریری آپ کے ماڈل سے رابطہ آسان بناتی ہے اور پرامپٹس بھیجتی ہے۔

پھر کچھ ایسی لائبریریاں بھی ہیں جو اعلیٰ سطح پر کام کرتی ہیں جیسے:

- **Langchain**۔ Langchain مشہور ہے اور Python کو سپورٹ کرتا ہے۔
- **Semantic Kernel**۔ Semantic Kernel مائیکروسافٹ کی لائبریری ہے جو C#، Python، اور Java زبانیں سپورٹ کرتی ہے۔

### 2. VS Code (یا کوئی پسندیدہ IDE)

آپ کے Microsoft Foundry پروجیکٹ میں آپ کے پسندیدہ IDE میں چلانے کے لیے تمام معلومات موجود ہیں۔ یہ مثال VS Code دکھائے گی

* اپنے فاؤنڈری پروجیکٹ میں ایک چیٹ ماڈل جیسے `gpt-4o-mini` تعینات کریں
* Overview صفحہ سے پروجیکٹ کا **endpoint** اور **API key** کاپی کریں
* انہیں محفوظ کرنے کے لیے ماحول کے متغیر بنائیں: `AZURE_INFERENCE_ENDPOINT` اور `AZURE_INFERENCE_CREDENTIAL` - بش، پاورشیل اور ونڈوز کمانڈ پرامپٹ میں نمونے دستیاب ہیں
* درکار dependencies انسٹال کریں: ```pip install azure-ai-inference```
* بنیادی نمونہ کوڈ کو .py فائل میں کاپی کریں
* جہاں کوڈ محفوظ ہے وہاں جائیں اور فائل چلائیں: ```python filename.py```

مت بھولیں کہ Azure AI Inference SDK استعمال کرتے ہوئے، آپ مختلف ماڈلز کے ساتھ آسانی سے تجربہ کر سکتے ہیں صرف کوڈ میں `model_name` کی قدر تبدیل کرکے۔

## ChatCompletions کے ساتھ ٹیکسٹ جنریٹ کریں

ٹیکسٹ جنریٹ کرنے کا طریقہ `ChatCompletionsClient` کلاس کا استعمال ہے۔
`samples/python/azure_ai_inference/basic.py` میں، کوڈ کے ریسپانس سیکشن میں، یوزر رول کو اپ ڈیٹ کریں اور مواد کے پیرامیٹر کو نیچے دیے گئے مطابق ترمیم کریں:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # اختیاری پیرامیٹرز
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

اپ ڈیٹ شدہ فائل چلائیں تاکہ آؤٹ پٹ دیکھ سکیں۔


## مختلف چیزوں کے لیے مختلف قسم کے پرامپٹس  

اب آپ نے دیکھا کہ کس طرح پرامپٹ استعمال کرکے متن تیار کیا جا سکتا ہے۔ آپ کے پاس ایک پروگرام بھی چل رہا ہے جسے آپ مختلف قسم کا متن بنانے کے لیے ترمیم اور تبدیل کر سکتے ہیں۔  

پرامپٹس کو ہر قسم کے کاموں کے لیے استعمال کیا جا سکتا ہے۔ مثال کے طور پر:  

- **کسی قسم کا متن پیدا کریں**۔ مثلاً، آپ نظم، کوئز کے سوالات وغیرہ بنا سکتے ہیں۔  
- **معلومات تلاش کریں**۔ آپ پرامپٹس استعمال کرکے معلومات دریافت کر سکتے ہیں، جیسا کہ مندرجہ ذیل مثال 'ویب ڈیولپمنٹ میں CORS کا کیا مطلب ہے؟'۔

1. ایک `prompt` ویریبل بنائیں اور نمونہ کوڈ کو درج ذیل کی طرح تبدیل کریں:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

   درجہ حرارت کو تبدیل کرنے کے لیے، آپ `temperature` پیرامیٹر استعمال کر سکتے ہیں۔ مثال کے طور پر، اگر آپ 0.5 درجہ حرارت استعمال کرنا چاہتے ہیں، تو آپ یوں کریں گے:

```python
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
            },
            {
                "role": "user",
                "content": new_prompt,
            },
        ],
        model=model_name,
        # اختیاری پیرامیٹرز
        temperature=0.5,
        max_tokens=1200,
        top_p=1.    
    )
```

   > نوٹ کریں، جتنا قدر 1.0 کے قریب ہو، نتیجہ اتنا ہی زیادہ متنوع ہوگا۔
