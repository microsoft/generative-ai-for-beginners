# ٹیکسٹ جنریشن ایپس بنائیں

آپ نے اس نصاب میں اب تک دیکھا ہے کہ کچھ بنیادی تصورات جیسے پرامپٹس اور یہاں تک کہ ایک مکمل شعبہ "پرامپٹ انجینئرنگ" کے نام سے موجود ہے۔ بہت سے ٹولز جیسے ChatGPT، Office 365، Microsoft Power Platform وغیرہ آپ کو پرامپٹس کے ذریعے کچھ حاصل کرنے میں مدد دیتے ہیں۔

اگر آپ اپنی ایپ میں ایسا تجربہ شامل کرنا چاہتے ہیں تو آپ کو پرامپٹس، کمپلیشنز جیسے تصورات کو سمجھنا ہوگا اور کام کرنے کے لیے کوئی لائبریری منتخب کرنی ہوگی۔ یہی کچھ آپ اس باب میں سیکھیں گے۔

## تعارف

اس باب میں آپ:

- openai لائبریری اور اس کے بنیادی تصورات کے بارے میں سیکھیں گے۔
- openai کا استعمال کرتے ہوئے ایک ٹیکسٹ جنریشن ایپ بنائیں گے۔
- پرامپٹ، ٹمپریچر اور ٹوکنز جیسے تصورات کو استعمال کرنا سیکھیں گے تاکہ ایک ٹیکسٹ جنریشن ایپ بنائی جا سکے۔

## سیکھنے کے مقاصد

اس سبق کے آخر میں آپ یہ کرنے کے قابل ہوں گے:

- وضاحت کر سکیں گے کہ ٹیکسٹ جنریشن ایپ کیا ہے۔
- openai کا استعمال کرتے ہوئے ایک ٹیکسٹ جنریشن ایپ بنا سکیں گے۔
- اپنی ایپ کو زیادہ یا کم ٹوکنز استعمال کرنے کے لیے ترتیب دے سکیں گے اور ساتھ ہی ٹمپریچر بھی بدل سکیں گے تاکہ مختلف آؤٹ پٹ حاصل ہو۔

## ٹیکسٹ جنریشن ایپ کیا ہے؟

عام طور پر جب آپ کوئی ایپ بناتے ہیں تو اس کا کوئی نہ کوئی انٹرفیس ہوتا ہے جیسے کہ:

- کمانڈ بیسڈ۔ کنسول ایپس عام طور پر وہ ایپس ہوتی ہیں جن میں آپ کمانڈ لکھتے ہیں اور وہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ بیسڈ ایپ ہے۔
- یوزر انٹرفیس (UI)۔ کچھ ایپس میں گرافیکل یوزر انٹرفیس (GUI) ہوتا ہے جہاں آپ بٹن کلک کرتے ہیں، ٹیکسٹ لکھتے ہیں، آپشنز منتخب کرتے ہیں وغیرہ۔

### کنسول اور UI ایپس کی حدود

اس کا موازنہ ایک کمانڈ بیسڈ ایپ سے کریں جس میں آپ کمانڈ لکھتے ہیں:

- **اس کی حدود ہیں**۔ آپ کوئی بھی کمانڈ نہیں لکھ سکتے، صرف وہی جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس کئی زبانیں سپورٹ کرتی ہیں، لیکن عام طور پر ایپ ایک خاص زبان کے لیے بنائی جاتی ہے، چاہے آپ مزید زبانیں شامل کر بھی لیں۔

### ٹیکسٹ جنریشن ایپس کے فائدے

تو ٹیکسٹ جنریشن ایپ کس طرح مختلف ہے؟

ٹیکسٹ جنریشن ایپ میں آپ کو زیادہ آزادی ملتی ہے، آپ کسی خاص کمانڈ یا زبان تک محدود نہیں ہوتے۔ اس کے بجائے، آپ قدرتی زبان میں ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ آپ ایک ایسے ڈیٹا سورس سے بات کر رہے ہوتے ہیں جو وسیع معلومات پر تربیت یافتہ ہے، جبکہ روایتی ایپ صرف ڈیٹا بیس تک محدود ہو سکتی ہے۔

### میں ٹیکسٹ جنریشن ایپ سے کیا بنا سکتا ہوں؟

آپ بہت سی چیزیں بنا سکتے ہیں۔ مثلاً:

- **چیٹ بوٹ**۔ ایک چیٹ بوٹ جو آپ کی کمپنی یا اس کی مصنوعات کے بارے میں سوالات کے جواب دے سکتا ہے۔
- **ہیلپر**۔ LLMs ٹیکسٹ کو خلاصہ کرنے، اس سے معلومات نکالنے، ریزیومے وغیرہ بنانے میں بہت اچھے ہیں۔
- **کوڈ اسسٹنٹ**۔ آپ جس لینگویج ماڈل کو استعمال کریں، اس کے مطابق آپ ایک کوڈ اسسٹنٹ بنا سکتے ہیں جو کوڈ لکھنے میں مدد دے۔ مثلاً، آپ GitHub Copilot یا ChatGPT کو کوڈ لکھنے میں مدد کے لیے استعمال کر سکتے ہیں۔

## میں کیسے شروع کروں؟

آپ کو LLM کے ساتھ انٹیگریٹ کرنے کا کوئی طریقہ تلاش کرنا ہوگا، جو عام طور پر دو طریقوں میں سے ایک ہوتا ہے:

- API استعمال کریں۔ یہاں آپ اپنی پرامپٹ کے ساتھ ویب ریکویسٹ بناتے ہیں اور جواب میں جنریٹڈ ٹیکسٹ حاصل کرتے ہیں۔
- لائبریری استعمال کریں۔ لائبریریز API کالز کو آسان بناتی ہیں اور استعمال میں سہولت دیتی ہیں۔

## لائبریریز/SDKs

LLMs کے ساتھ کام کرنے کے لیے کچھ مشہور لائبریریز ہیں جیسے:

- **openai**، یہ لائبریری آپ کے ماڈل سے جڑنے اور پرامپٹس بھیجنے کو آسان بناتی ہے۔

اس کے علاوہ کچھ لائبریریز زیادہ اعلیٰ سطح پر کام کرتی ہیں جیسے:

- **Langchain**۔ Langchain کافی مشہور ہے اور Python کو سپورٹ کرتا ہے۔
- **Semantic Kernel**۔ Semantic Kernel مائیکروسافٹ کی لائبریری ہے جو C#, Python اور Java کو سپورٹ کرتی ہے۔

## ChatCompletions کے ساتھ ٹیکسٹ جنریٹ کریں

ٹیکسٹ جنریٹ کرنے کا طریقہ یہ ہے کہ `ChatCompletionsClient` کلاس استعمال کریں۔
`samples/python/azure_ai_inference/basic.py` میں، response سیکشن میں، کوڈ میں user رول کے content پیرامیٹر کو نیچے دی گئی ویلیو سے بدلیں:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

## مختلف قسم کے پرامپٹس، مختلف کاموں کے لیے

اب آپ نے دیکھ لیا ہے کہ پرامپٹ کے ذریعے کس طرح ٹیکسٹ تیار کیا جا سکتا ہے۔ آپ کے پاس ایک پروگرام بھی ہے جسے آپ اپنی مرضی کے مطابق بدل سکتے ہیں اور مختلف قسم کا ٹیکسٹ بنا سکتے ہیں۔

پرامپٹس کو مختلف کاموں کے لیے استعمال کیا جا سکتا ہے۔ مثلاً:

- **کسی خاص قسم کا ٹیکسٹ تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔
- **معلومات تلاش کرنا**۔ آپ پرامپٹس کے ذریعے معلومات تلاش کر سکتے ہیں، جیسے یہ مثال: 'ویب ڈیولپمنٹ میں CORS کا کیا مطلب ہے؟'
- **کوڈ تیار کرنا**۔ آپ پرامپٹس کے ذریعے کوڈ بھی تیار کر سکتے ہیں، مثلاً ای میل کی تصدیق کے لیے ریگولر ایکسپریشن بنانا یا پھر پورا پروگرام تیار کرنا، جیسے کہ ویب ایپ۔

## مشق - ایک نسخہ جنریٹر بنائیں

اب جب کہ ہم نے ایک منظرنامہ آزما لیا ہے، آئیے اس منظرنامے کے مطابق کوڈ لکھتے ہیں۔ اس کے لیے، یہ اقدامات کریں:

1. موجودہ فائل کو نقطہ آغاز کے طور پر استعمال کریں
1. ایک `prompt` متغیر بنائیں اور نمونہ کوڈ کو نیچے دیے گئے طریقے کے مطابق تبدیل کریں:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

> [!NOTE] یاد رکھیں، آپ کا LLM غیر متعین ہے، اس لیے ہر بار پروگرام چلانے پر مختلف نتائج مل سکتے ہیں۔

زبردست، اب دیکھتے ہیں کہ ہم چیزوں کو کیسے بہتر بنا سکتے ہیں۔ چیزوں کو بہتر بنانے کے لیے، ہم چاہتے ہیں کہ کوڈ لچکدار ہو، تاکہ اجزاء اور ترکیبوں کی تعداد کو آسانی سے بہتر اور تبدیل کیا جا سکے۔


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
