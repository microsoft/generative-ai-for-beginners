# ٹیکسٹ جنریشن ایپس بنائیں

آپ نے اب تک اس نصاب میں دیکھا ہے کہ بنیادی تصورات ہوتے ہیں جیسے پرامپٹس اور یہاں تک کہ ایک پوری ڈسپلن ہے جسے "پرامپٹ انجینئرنگ" کہا جاتا ہے۔ بہت سے ٹولز جن سے آپ بات چیت کر سکتے ہیں جیسے ChatGPT، Office 365، Microsoft Power Platform اور دیگر، آپ کو کسی کام کو پورا کرنے کے لیے پرامپٹس استعمال کرنے کی سہولت دیتے ہیں۔

اگر آپ ایسی تجربہ ایپ میں شامل کرنا چاہتے ہیں، تو آپ کو پرامپٹس، مکمل کرنے جیسے تصورات کو سمجھنا ہوگا اور ایک لائبریری منتخب کرنی ہوگی جس کے ساتھ کام کیا جا سکے۔ یہی آپ اس باب میں سیکھیں گے۔

## تعارف

اس باب میں، آپ:

- اوپن اے آئی لائبریری اور اس کے بنیادی تصورات کے بارے میں سیکھیں گے۔
- اوپن اے آئی کا استعمال کرتے ہوئے ایک ٹیکسٹ جنریشن ایپ بنائیں گے۔
- یہ سمجھیں گے کہ پرامپٹ، ٹیمپریچر، اور ٹوکنز جیسے تصورات کو کس طرح استعمال کیا جا سکتا ہے تاکہ ٹیکسٹ جنریشن ایپ بنائی جا سکے۔

## سیکھنے کے اہداف

اس سبق کے آخر میں، آپ قابل ہوں گے کہ:

- وضاحت کریں کہ ٹیکسٹ جنریشن ایپ کیا ہے۔
- اوپن اے آئی کا استعمال کرتے ہوئے ٹیکسٹ جنریشن ایپ بنائیں۔
- اپنی ایپ کو زیادہ یا کم ٹوکنز استعمال کرنے کے لیے ترتیب دیں اور ٹیمپریچر بھی تبدیل کریں، تاکہ مختلف نتائج حاصل کیے جا سکیں۔

## ٹیکسٹ جنریشن ایپ کیا ہے؟

عام طور پر جب آپ ایپ بناتے ہیں تو اس میں کچھ قسم کا انٹرفیس ہوتا ہے، جیسے درج ذیل:

- کمانڈ پر مبنی۔ کنسول ایپس عام ایپس ہوتی ہیں جہاں آپ کمانڈ لکھتے ہیں اور یہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ پر مبنی ایپ ہے۔
- صارف انٹرفیس (UI)۔ بعض ایپس میں گرافیکل یوزر انٹرفیس (GUIs) ہوتا ہے جہاں آپ بٹن کلک کرتے ہیں، ٹیکسٹ داخل کرتے ہیں، اختیارات منتخب کرتے ہیں اور مزید۔

### کنسول اور UI ایپس محدود ہوتی ہیں

اسے ایک کمانڈ پر مبنی ایپ سے موازنہ کریں جہاں آپ کمانڈ ٹائپ کرتے ہیں:

- **یہ محدود ہے**۔ آپ کوئی بھی کمانڈ نہیں ٹائپ کر سکتے، صرف وہی جنہیں ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس کئی زبانوں کی حمایت کرتی ہیں، لیکن بطور ڈیفالٹ ایپ خاص زبان کے لیے بنتی ہے، اگرچہ آپ مزید زبانوں کی حمایت شامل کر سکتے ہیں۔

### ٹیکسٹ جنریشن ایپس کے فوائد

تو ٹیکسٹ جنریشن ایپ کس طرح مختلف ہے؟

ٹیکسٹ جنریشن ایپ میں آپ کے پاس زیادہ لچک ہوتی ہے، آپ کمانڈز کے ایک سیٹ یا مخصوص ان پٹ زبان تک محدود نہیں ہوتے۔ اس کے بجائے، آپ قدرتی زبان استعمال کر کے ایپ کے ساتھ بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ چونکہ آپ ایک ایسے ڈیٹا سورس سے بات کر رہے ہیں جسے وسیع معلومات کے ذخیرے پر تربیت دی گئی ہے، جبکہ روایتی ایپ ممکنہ طور پر ڈیٹا بیس میں موجود مواد تک محدود ہوسکتی ہے۔

### میں ٹیکسٹ جنریشن ایپ سے کیا بنا سکتا ہوں؟

آپ بہت سی چیزیں بنا سکتے ہیں۔ مثال کے طور پر:

- **چیٹ بوٹ**۔ ایک چیٹ بوٹ جو سوالات کے جواب دیتا ہے، جیسے کہ آپ کی کمپنی اور اس کی مصنوعات کے بارے میں، اچھا میل ہو سکتا ہے۔
- **مددگار**۔ ایل ایل ایمز متن کو خلاصہ کرنے، متعلقہ معلومات نکالنے، ریزیومے بنانے جیسے کاموں میں بہترین ہوتے ہیں۔
- **کوڈ اسسٹنٹ**۔ جس زبان کے ماڈل کا آپ استعمال کرتے ہیں، اس کے مطابق آپ کوڈ اسسٹنٹ بنا سکتے ہیں جو کوڈ لکھنے میں مدد دے۔ مثال کے طور پر، آپ GitHub Copilot یا ChatGPT جیسا پروڈکٹ استعمال کر سکتے ہیں تاکہ کوڈ لکھنے میں مدد ملے۔

## میں کیسے شروع کر سکتا ہوں؟

تو، آپ کو ایل ایل ایم کے ساتھ انٹیگریٹ کرنے کا کوئی طریقہ تلاش کرنا ہوگا جس کے لیے عام طور پر درج ذیل دو طریقے ہوتے ہیں:

- لائبریری استعمال کریں۔ لائبریریاں API کالز کو انکیپسولیٹ کرتی ہیں اور انہیں استعمال کرنا آسان بناتی ہیں۔

### 2. VS Code (یا کوئی بھی پسندیدہ IDE)

اپنے Microsoft Foundry پروجیکٹ سے آپ کے پاس وہ تمام معلومات ہیں جو آپ کو اپنی پسندیدہ IDE میں چلانے کے لیے چاہیے۔ یہ مثال VS Code دکھائے گی۔

* پروجیکٹ کا **endpoint** اور **API key** Overview صفحے سے کاپی کریں۔
* انہیں ذخیرہ کرنے کے لیے ماحول کے متغیرات بنائیں: `AZURE_INFERENCE_ENDPOINT` اور `AZURE_INFERENCE_CREDENTIAL` - نمونے bash، powershell اور ونڈوز کمانڈ پرامپٹ میں دستیاب ہیں۔

* بنیادی نمونہ کوڈ کو .py فائل میں کاپی کریں۔
* جہاں کوڈ محفوظ کیا ہے وہاں جا کر فائل چلائیں: `python filename.py`

## ChatCompletions کے ساتھ ٹیکسٹ جنریٹ کریں

ٹیکسٹ جنریٹ کرنے کا طریقہ `ChatCompletionsClient` کلاس کا استعمال ہے۔
`samples/python/azure_ai_inference/basic.py` میں، جواب والے کوڈ کے سیکشن میں، صارف رول کو اپ ڈیٹ کریں اور content پیرامیٹر کو نیچے دیے گئے مطابق تبدیل کریں:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # اختیاری پیرامیٹرز
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

اپ ڈیٹ شدہ فائل چلائیں تاکہ آؤٹ پٹ دیکھ سکیں۔


## مختلف چیزوں کے لیے مختلف اقسام کے پرامپٹس  

اب آپ نے دیکھا کہ پرامپٹ کا استعمال کرتے ہوئے متن کیسے تخلیق کیا جاتا ہے۔ آپ کے پاس ایک پروگرام بھی چل رہا ہے جسے آپ مختلف قسم کے متن تخلیق کرنے کے لیے تبدیل اور ترمیم کر سکتے ہیں۔  

پرامپٹس مختلف قسم کے کاموں کے لیے استعمال کیے جا سکتے ہیں۔ مثال کے طور پر:  

- **متن کی ایک قسم تخلیق کریں**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تخلیق کر سکتے ہیں۔  
- **معلومات تلاش کریں**۔ آپ پرامپٹس کا استعمال معلومات تلاش کرنے کے لیے کر سکتے ہیں جیسے درج ذیل مثال: 'ویب ڈویلپمنٹ میں CORS کا کیا مطلب ہے؟'۔

## مشق - ایک نسخہ جنریٹر بنائیں

چونکہ ہم نے ایک منظر پیش کیا ہے، آئیے اس منظر سے میل کھاتا ہوا کوڈ لکھیں۔ ایسا کرنے کے لیے، درج ذیل مراحل پر عمل کریں:

1. موجودہ فائل کو نقطہ آغاز کے طور پر استعمال کریں
1. ایک `prompt` متغیر بنائیں اور نمونہ کوڈ کو نیچے کی طرح تبدیل کریں:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. آئیں کوڈ کو درج ذیل طریقے سے تبدیل کریں:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

        ```output
        Feel free to omit the optional ingredients if you prefer or if you already have them on hand. Enjoy your delicious strawberry milkshake!
        ```
        
- **درجہ حرارت (temperature) کے ساتھ تجربہ کرنا**۔ درجہ حرارت ایسی چیز ہے جس کا ہم نے اب تک ذکر نہیں کیا لیکن یہ ہمارے پروگرام کی کارکردگی کے لیے ایک اہم سیاق و سباق ہے۔ جتنا زیادہ درجہ حرارت ہوگا اتنا زیادہ غیر متوقع آؤٹ پٹ ہوگا۔ اس کے برعکس، جتنا کم درجہ حرارت ہوگا اتنا زیادہ پیش گوئی والا آؤٹ پٹ ہوگا۔ غور کریں کہ آیا آپ اپنے آؤٹ پٹ میں تبدیلی چاہتے ہیں یا نہیں۔

   درجہ حرارت کو تبدیل کرنے کے لیے، آپ `temperature` پیرامیٹر استعمال کر سکتے ہیں۔ مثال کے طور پر، اگر آپ 0.5 درجہ حرارت استعمال کرنا چاہتے ہیں، تو آپ یوں کریں گے:

```python
    response = client.complete(

                "content": new_prompt,
            },
        ],
        model=model_name,
        # اختیاری پیرامیٹرز
        temperature=0.5,
        max_tokens=1200,
        top_p=1.    
    )
```
