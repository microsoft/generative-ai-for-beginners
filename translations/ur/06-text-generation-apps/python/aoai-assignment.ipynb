## متن تخلیق کرنے والی ایپ کیا ہے؟

عموماً جب آپ ایپ بناتے ہیں، تو اس کا کوئی نہ کوئی انٹرفیس ہوتا ہے مثلاً درج ذیل:

- کمانڈ پر مبنی۔ کنسول ایپس عام ایپس ہوتی ہیں جہاں آپ کمانڈ ٹائپ کرتے ہیں اور یہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ پر مبنی ایپ ہے۔
- یوزر انٹرفیس (UI). کچھ ایپس میں گرافیکل یوزر انٹرفیس (GUIs) ہوتے ہیں جہاں آپ بٹن کلک کرتے ہیں، متن داخل کرتے ہیں، آپشن منتخب کرتے ہیں اور مزید۔

### کنسول اور UI ایپس محدود ہوتی ہیں

اسے ایک کمانڈ پر مبنی ایپ سے موازنہ کریں جہاں آپ کمانڈ ٹائپ کرتے ہیں:

- **یہ محدود ہوتی ہیں**۔ آپ کوئی بھی کمانڈ ٹائپ نہیں کر سکتے، صرف وہی جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس بہت سی زبانیں سپورٹ کرتی ہیں، مگر عام طور پر ایپ ایک خاص زبان کے لیے بنائی جاتی ہے، چاہے آپ مزید زبانیں شامل کر سکیں۔

### متن تخلیق کرنے والی ایپس کے فائدے

تو متن تخلیق کرنے والی ایپ کس طرح مختلف ہے؟

متن تخلیق کرنے والی ایپ میں، آپ کو زیادہ آزادی حاصل ہوتی ہے، آپ کسی مخصوص کمانڈ مجموعه یا مخصوص ان پٹ زبان تک محدود نہیں ہوتے۔ بلکہ، آپ قدرتی زبان استعمال کر کے ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ چونکہ آپ پہلے ہی ایک ایسے ڈیٹا سورس سے بات چیت کر رہے ہیں جو وسیع مقدار میں معلومات پر تربیت یافتہ ہے، جبکہ روایتی ایپ ممکن ہے کہ ڈیٹا بیس میں موجود مواد تک محدود ہو۔

### متن تخلیق کرنے والی ایپ سے کیا بنا سکتا ہوں؟

آپ بہت سی چیزیں بنا سکتے ہیں۔ مثلاً:

- **چیٹ بوٹ**۔ ایک چیٹ بوٹ جو سوالات کے جواب دے مثلاً آپ کی کمپنی اور اس کی مصنوعات کے بارے میں، ایک اچھا انتخاب ہو سکتا ہے۔

- **کوڈ اسسٹنٹ**۔ آپ جو زبان ماڈل استعمال کرتے ہیں اس کے حساب سے، آپ کوڈ لکھنے میں مدد دینے والا کوڈ اسسٹنٹ بنا سکتے ہیں۔ مثال کے طور پر، آپ GitHub Copilot اور ChatGPT دونوں استعمال کر کے کوڈ لکھنے میں مدد حاصل کر سکتے ہیں۔

### Azure کی ترتیب سیٹ کریں

اگر آپ Azure OpenAI استعمال کر رہے ہیں، تو یہاں ترتیب دینے کا طریقہ ہے۔ Responses API Azure OpenAI **v1 endpoint** سے فراہم کی جاتی ہے، تو ہم `OpenAI` کلائنٹ کو `<your-endpoint>/openai/v1/` پر پوائنٹ کرتے ہیں:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

اوپر ہم درج ذیل سیٹ کر رہے ہیں:

- `api_key`، یہ آپ کی API کی ہے جو Azure Portal میں ملی ہے۔
- `base_url`, یہ آپ کا Azure OpenAI endpoint ہے جس کے ساتھ `/openai/v1/` جڑا ہوا ہے۔ آپ endpoint Azure Portal میں اپنی API کی کے پاس دیکھ سکتے ہیں۔ v1 endpoint استعمال کرنے کا مطلب ہے کہ آپ کو `api_version` پاس کرنے کی ضرورت نہیں۔
- `deployment`, یہ اس ماڈل ڈپلائمنٹ کا نام ہے جو آپ نے Foundry پورٹل میں بنایا ہے۔

> [!NOTE]
> `os.environ` ایک میپنگ ہے جو ماحول کی متغیرات پڑھتی ہے۔ آپ اسے `AZURE_OPENAI_API_KEY` اور `AZURE_OPENAI_ENDPOINT` جیسی متغیرات پڑھنے کے لیے استعمال کر سکتے ہیں۔

## متن تخلیق کریں

متن تخلیق کرنے کا طریقہ `responses.create` میتھڈ استعمال کرنا ہے۔ مثال دیکھیں:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

اوپر والے کوڈ میں، ہم ایک response object بناتے ہیں اور ماڈل اور پرامپٹ پاس کرتے ہیں۔ پھر ہم پیدا کردہ متن پرنٹ کرتے ہیں۔

### چیٹ کے جوابات

Responses API ایک ٹرن کی متن تخلیق اور ملٹی ٹرن چیٹ بوٹس دونوں کے لیے موزوں ہے - آپ صرف `input` فہرست میں مزید پیغامات فراہم کرتے ہیں تاکہ گفتگو بن سکے:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],

```

## مشق - آپ کی پہلی متن تخلیق کرنے والی ایپ

1. ایک ورچوئل ماحول بنائیں اور openai انسٹال کریں:

  > [!NOTE] اگر آپ یہ نوٹ بک Codespaces پر یا Devcontainer کے اندر چلا رہے ہیں تو یہ قدم ضروری نہیں ہے

> [!NOTE]
> اگر آپ ونڈوز استعمال کر رہے ہیں تو `source venv/bin/activate` کی بجائے `venv\Scripts\activate` ٹائپ کریں۔

> [!NOTE]
> اپنا Azure OpenAI کلید تلاش کرنے کے لیے https://portal.azure.com/ پر جائیں، `Open AI` تلاش کریں، `Open AI resource` منتخب کریں، پھر `Keys and Endpoint` منتخب کریں اور `Key 1` کی قدر کو کاپی کریں۔


1۔ ایک *app.py* فائل بنائیں اور اسے مندرجہ ذیل کوڈ دیں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    آپ کو مندرجہ ذیل جیسا آؤٹ پٹ دیکھنا چاہیے:

    ```output
     بہت اداس _____.

    ایک دفعہ کا ذکر ہے کہ ایک بہت اداس ماہی پری تھی۔
    ```


## مختلف قسم کے پرامپٹس، مختلف چیزوں کے لیے

پرامپٹس مختلف قسم کے کاموں کے لیے استعمال کیے جا سکتے ہیں۔ مثال کے طور پر:

- **کسی قسم کا متن تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔
- **معلومات حاصل کرنا**۔ آپ معلومات حاصل کرنے کے لیے پرامپٹس استعمال کر سکتے ہیں، جیسے درج ذیل مثال 'ویب ڈویلپمنٹ میں CORS کا کیا مطلب ہے؟'۔
- **کوڈ تیار کرنا**۔ آپ کوڈ بنانے کے لیے پرامپٹس استعمال کر سکتے ہیں، مثال کے طور پر ای میلز کی تصدیق کے لیے ریگولر ایکسپریشن تیار کرنا یا پورا پروگرام تیار کرنا، جیسے ایک ویب ایپ؟
