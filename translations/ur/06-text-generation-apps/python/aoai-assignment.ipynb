## متن جنریشن ایپ کیا ہے؟

عام طور پر جب آپ ایک ایپ بناتے ہیں تو اس کا کچھ انٹرفیس ہوتا ہے جیسا کہ مندرجہ ذیل:

- کمانڈ پر مبنی۔ کنسول ایپس عام ایپس ہیں جہاں آپ کمانڈ ٹائپ کرتے ہیں اور یہ کوئی کام انجام دیتی ہے۔ مثال کے طور پر، `git` ایک کمانڈ پر مبنی ایپ ہے۔
- صارف انٹرفیس (UI)۔ کچھ ایپس میں گرافیکل صارف انٹرفیس (GUIs) ہوتی ہیں جہاں آپ بٹنز کلک کرتے ہیں، متن داخل کرتے ہیں، اختیارات کا انتخاب کرتے ہیں اور مزید۔

### کنسول اور UI ایپس محدود ہیں

اس کا موازنہ ایک کمانڈ پر مبنی ایپ سے کریں جہاں آپ کمانڈ ٹائپ کرتے ہیں:

- **یہ محدود ہیں**۔ آپ کسی بھی کمانڈ کو نہیں ٹائپ کر سکتے، صرف وہ جو ایپ سپورٹ کرتی ہے۔
- **زبان مخصوص**۔ کچھ ایپس متعدد زبانیں سپورٹ کرتی ہیں، لیکن عمومی طور پر ایپ ایک مخصوص زبان کے لیے بنائی جاتی ہے، پھر بھی آپ مزید زبان کی سپورٹ شامل کر سکتے ہیں۔

### متن جنریشن ایپس کے فوائد

تب متن جنریشن ایپ مختلف کیسے ہے؟

متن جنریشن ایپ میں آپ کے پاس زیادہ لچک ہوتی ہے، آپ کمانڈز کے کسی مخصوص سیٹ یا کسی مخصوص زبان کیلئے محدود نہیں ہوتے۔ اس کے بجائے، آپ قدرتی زبان استعمال کر کے ایپ سے بات چیت کر سکتے ہیں۔ ایک اور فائدہ یہ ہے کہ چونکہ آپ ایسے ڈیٹا سورس کے ساتھ تعامل کر رہے ہیں جو وسیع معلومات پر مشتمل تربیت حاصل کر چکا ہے، جب کہ روایتی ایپ ڈیٹا بیس میں موجود مواد تک محدود ہو سکتی ہے۔

### متن جنریشن ایپ کے ساتھ کیا بنا سکتا ہوں؟

بہت سی چیزیں آپ بنا سکتے ہیں۔ مثلاً:

- **ایک چیٹ بوٹ**۔ ایک چیٹ بوٹ جو کمپنی اور اس کی مصنوعات کے متعلق سوالات کے جوابات دے، اچھا انتخاب ہو سکتا ہے۔

- **کوڈ اسسٹنٹ**۔ جس زبان کا ماڈل آپ استعمال کرتے ہیں اس کے مطابق، آپ کوڈ لکھنے میں مدد دینے والا کوڈ اسسٹنٹ بنا سکتے ہیں۔ مثال کے طور پر، آپ GitHub Copilot یا ChatGPT جیسا پروڈکٹ استعمال کر کے کوڈ لکھنے میں مدد حاصل کر سکتے ہیں۔

### Azure کی ترتیب سیٹ کریں

اگر آپ Azure OpenAI استعمال کر رہے ہیں تو یہاں ترتیب سیٹ کرنے کا طریقہ ہے۔ Responses API Azure OpenAI کے **v1 endpoint** سے فراہم کی جاتی ہے، اس لیے ہم `OpenAI` کلائنٹ کو `<your-endpoint>/openai/v1/` پر پوائنٹ کرتے ہیں:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

اوپر ہم درج ذیل سیٹ کر رہے ہیں:

- `api_key`، یہ آپ کی Azure Portal میں پائی جانے والی API کلید ہے۔
- `base_url`، یہ آپ کے Azure OpenAI اینڈپوائنٹ کا وہ URL ہے جس کے آخر میں `/openai/v1/` لگا ہوا ہے۔ آپ اینڈپوائنٹ Azure Portal میں اپنی API کلید کے ساتھ دیکھ سکتے ہیں۔ v1 اینڈپوائنٹ استعمال کرنے کا مطلب ہے کہ آپ کو `api_version` پارامیٹر منتقل کرنے کی ضرورت نہیں ہے۔
- `deployment`، یہ اس ماڈل کی تعیناتی کا نام ہے جو آپ نے Foundry پورٹل میں بنایا ہے۔

> [!NOTE]
> `os.environ` ایک میپنگ ہے جو ماحول کے متغیرات پڑھتی ہے۔ آپ اسے `AZURE_OPENAI_API_KEY` اور `AZURE_OPENAI_ENDPOINT` جیسے متغیرات پڑھنے کے لیے استعمال کر سکتے ہیں۔

## متن تیار کریں

متن تیار کرنے کا طریقہ `responses.create` طریقہ استعمال کرنا ہے۔ یہاں ایک مثال ہے:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

اوپر دیے گئے کوڈ میں ہم ایک response آبجیکٹ بناتے ہیں، ماڈل اور پرامپٹ پاس کرتے ہیں۔ پھر ہم تیار شدہ متن پرنٹ کرتے ہیں۔

### چیٹ کے جواب

Responses API ایک سنگل ٹرن متن جنریشن اور ملٹی ٹرن چیٹ بوٹس دونوں کے لیے موزوں ہے - آپ بس `input` لسٹ میں مزید پیغامات فراہم کریں تاکہ بات چیت کی تشکیل ہو:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

اس خصوصیت پر مزید تفصیل آنے والے باب میں دی جائے گی۔

## مشق - آپ کی پہلی متن جنریشن ایپ

اب جب ہم نے سیکھ لیا کہ Azure OpenAI سروس کو کیسے سیٹ اپ اور ترتیب دیا جائے، اب وقت ہے کہ آپ کی پہلی متن جنریشن ایپ بنائیں۔ اپنی ایپ بنانے کے لیے، درج ذیل مراحل پر عمل کریں:


1. ایک ورچوئل ماحول بنائیں اور openai انسٹال کریں:

  > [!NOTE] اگر آپ یہ نوٹ بک Codespaces پر یا Devcontainer کے اندر چلاتے ہیں تو یہ مرحلہ ضروری نہیں ہے


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> اگر آپ ونڈوز استعمال کر رہے ہیں تو `source venv/bin/activate` کی بجائے `venv\Scripts\activate` ٹائپ کریں۔

> [!NOTE]
> اپنا Azure OpenAI کلید تلاش کرنے کے لیے https://portal.azure.com/ پر جائیں، `Open AI` تلاش کریں، `Open AI resource` منتخب کریں، پھر `Keys and Endpoint` منتخب کریں اور `Key 1` کی ویلیو کو کاپی کریں۔


1. ایک *app.py* فائل بنائیں اور اسے مندرجہ ذیل کوڈ دیں:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    آپ کو درج ذیل کی طرح نتیجہ دیکھنا چاہئے:

    ```output
     بہت ناخوش _____۔

    ایک بار کا ذکر ہے ایک بہت ناخوش مرمڈ تھی۔
    ```


## مختلف چیزوں کے لیے مختلف قسم کے پرامپٹس

اب آپ نے دیکھا کہ پرامپٹ استعمال کرکے متن کیسے تیار کیا جاتا ہے۔ آپ کے پاس ایک پروگرام بھی چل رہا ہے جسے آپ تبدیل کر کے مختلف قسم کے متن تیار کر سکتے ہیں۔

پرامپٹس مختلف قسم کے کاموں کے لیے استعمال کیے جا سکتے ہیں۔ مثال کے طور پر:

- **کسی قسم کا متن تیار کرنا**۔ مثال کے طور پر، آپ نظم، کوئز کے سوالات وغیرہ تیار کر سکتے ہیں۔
- **معلومات کی تلاش**۔ آپ پرامپٹس کو معلومات حاصل کرنے کے لیے استعمال کر سکتے ہیں جیسا کہ مثال میں ’ویب ڈویلپمنٹ میں CORS کا کیا مطلب ہے؟‘۔
- **کوڈ تیار کرنا**۔ آپ پرامپٹس کو کوڈ تیار کرنے کے لیے استعمال کر سکتے ہیں، مثلاً ای میلز کی تصدیق کے لیے ریگولر ایکسپریشن ڈیویلپ کرنا یا پورا پروگرام تیار کرنا، جیسے ویب ایپ؟

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
