### 2. Tuo apukirjastot ja luo tunnistetiedot


```python
import os
from openai import OpenAI
import numpy as np
from dotenv import load_dotenv
load_dotenv()

#validate data inside .env file

# The Responses API is served from the Azure OpenAI (Microsoft Foundry) v1 endpoint,
# so we point the OpenAI client at <your-endpoint>/openai/v1/ (no api_version needed).
endpoint = os.environ['AZURE_OPENAI_ENDPOINT']
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{endpoint.rstrip('/')}/openai/v1/",
  )
```

### 3. Oikean mallin löytäminen  
Mallit kuten GPT-4o ja GPT-4o mini voivat ymmärtää ja tuottaa luonnollista kieltä. Azure OpenAI Microsoft Foundryssä tarjoaa valikoiman mallikykyjä, joista jokaisella on erilaiset teho- ja nopeustasot, jotka sopivat erilaisiin tehtäviin. 

[Azure OpenAI Microsoft Foundryn malleissa](https://learn.microsoft.com/en-us/azure/ai-foundry/foundry-models/concepts/models-sold-directly-by-azure?WT.mc_id=academic-105485-koreyst)



```python
# Select the deployment name configured in your .env file
model = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

#### Miten ohjata suurta kielimallia  
"Kaikista sisääntulotiedoista suurimpaan kielimalliin vaikuttavin on ylivoimaisesti tekstikehote(1).

Suuria kielimalleja voi ohjata tuottamaan vastauksia muutamalla tavalla:

- Ohjeistus: Kerro mallille, mitä haluat
- Viimeistely: Saanut mallin täydentämään haluamasi alun
- Demonstraatio: Näytä mallille, mitä haluat, joko:
- Muutama esimerkki kehotteessa
- Satoja tai tuhansia esimerkkejä hienosäätökoulutusdatassa



#### Kehotteiden luomiselle on kolme perusohjetta:

**Näytä ja kerro**. Tee selväksi, mitä haluat joko ohjeiden, esimerkkien tai molempien yhdistelmällä. Jos haluat, että malli luokittelee listan aakkosjärjestykseen tai luokittelee kappaleen tunnesävyn mukaan, näytä sille, että se on mitä haluat.

**Tarjoa laadukasta dataa**. Jos yrität rakentaa luokittelijaa tai saada mallin seuraamaan kaavaa, varmista, että esimerkkejä on riittävästi. Tarkista esimerkkisi huolellisesti — malli on yleensä tarpeeksi älykäs tunnistamaan peruskirjoitusvirheet ja antamaan niihin vastauksen, mutta se voi myös olettaa, että ne ovat tarkoituksellisia ja se voi vaikuttaa vastaukseen.

**Tarkista asetuksesi.** Temperature- ja top_p-asetukset ohjaavat, kuinka määrätietoisesti malli tuottaa vastauksen. Jos pyydät vastausta, jossa on vain yksi oikea vastaus, aseta nämä alhaisemmiksi. Jos taas haluat monipuolisempia vastauksia, aseta ne korkeammiksi. Yleisin virhe näissä asetuksissa on olettaa, että ne kontrolloivat "älykkyyttä" tai "luovuutta".

## Upotukset
Tässä osiossa näytetään, miten haetaan upotuksia ja löydetään samankaltaisuuksia sanojen, lauseiden ja dokumenttien välillä. Jotta voit suorittaa seuraavat muistiinpanot, sinun täytyy ottaa käyttöön malli, joka käyttää `text-embedding-ada-002` -perusmallia, ja asettaa sen käyttöönoton nimi .env-tiedostoon muuttujan `AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT` avulla.


### Malliluokitus - upotusmallin valinta

**Malliluokitus**: {family} - {capability} - {input-type} - {identifier}  

{family}     --> text-embedding  (upotusten perhe)  
{capability} --> ada             (kaikki muut upotusmallit poistuvat käytöstä vuonna 2024)  
{input-type} --> n/a             (määritetään vain haku malleille)  
{identifier} --> 002             (versio 002)  

model = 'text-embedding-ada-002'


  > [!NOTE] Seuraava vaihe ei ole tarpeen, jos ajat tätä muistikirjaa Codespacesissa tai Devcontainerissa


```python
# Dependencies for embeddings_utils
%pip install matplotlib plotly scikit-learn pandas
```

```python
def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
text = 'the quick brown fox jumped over the lazy dog'
model= os.environ['AZURE_OPENAI_EMBEDDINGS_DEPLOYMENT']
client.embeddings.create(input='[text]', model=model).data[0].embedding
```

```python

# compare several words
automobile_embedding  = client.embeddings.create(input='automobile', model=model).data[0].embedding
vehicle_embedding     = client.embeddings.create(input='vehicle', model=model).data[0].embedding
dinosaur_embedding    = client.embeddings.create(input='dinosaur', model=model).data[0].embedding
stick_embedding       = client.embeddings.create(input='stick', model=model).data[0].embedding

print(cosine_similarity(automobile_embedding, vehicle_embedding))
print(cosine_similarity(automobile_embedding, dinosaur_embedding))
print(cosine_similarity(automobile_embedding, stick_embedding))
```
