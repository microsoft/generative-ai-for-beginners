## Mikä on tekstintuottosovellus?

Tavallisesti, kun rakennat sovelluksen, siinä on jonkinlainen käyttöliittymä kuten seuraava:

- Komentorivipohjainen. Konsolisovellukset ovat tyypillisiä sovelluksia, joissa kirjoitat komennon ja se suorittaa tehtävän. Esimerkiksi `git` on komentopohjainen sovellus.
- Käyttöliittymä (UI). Joissakin sovelluksissa on graafisia käyttöliittymiä (GUI), joissa klikkaat painikkeita, syötät tekstiä, valitset vaihtoehtoja ja muuta.

### Konsoli- ja käyttöliittymäsovelluksilla on rajoituksensa

Vertaile sitä komentopohjaiseen sovellukseen, jossa kirjoitat komennon: 

- **Se on rajattu**. Et voi kirjoittaa mitä tahansa komentoa, vain niitä, joita sovellus tukee.
- **Kieli on rajattu**. Jotkut sovellukset tukevat monia kieliä, mutta oletuksena sovellus on rakennettu tietylle kielelle, vaikka voit lisätä lisää kielitukea.

### Tekstintuottosovellusten edut

Miten sitten tekstintuottosovellus on erilainen?

Tekstintuottosovelluksessa sinulla on enemmän joustavuutta, et ole rajoitettu joukkoon komentoja tai tiettyyn syötekieleen. Sen sijaan voit käyttää luonnollista kieltä kommunikoidaksesi sovelluksen kanssa. Toinen etu on, että koska olet jo vuorovaikutuksessa tietolähteen kanssa, joka on koulutettu valtavan tietomäärän pohjalta, perinteinen sovellus saattaa olla rajoitettu tietokannan sisältöön.

### Mitä voin rakentaa tekstintuottosovelluksella?

Voit rakentaa monenlaisia asioita. Esimerkiksi:

- **Chatbotin**. Chatbot, joka vastaa kysymyksiin aiheista, kuten yrityksestäsi ja sen tuotteista, voisi olla hyvä ratkaisu.
- **Avustajan**. LLM-mallit ovat loistavia asioissa kuten tekstin tiivistäminen, tekstistä saatavien oivallusten hakeminen, tekstin tuotanto kuten ansioluettelot ja muuta.
- **Koodiavustajan**. Käyttämästäsi kielimallista riippuen voit rakentaa koodiavustajan, joka auttaa sinua kirjoittamaan koodia. Esimerkiksi voit käyttää GitHub Copilotia sekä ChatGPT:tä auttamaan koodin kirjoittamisessa.

## Tuota tekstiä

Tekstin tuottamiseen käytetään `responses.create`-metodia. Tässä esimerkki:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Ylläolevassa koodissa luomme vastausobjektin ja annamme sen käyttämään haluamaamme mallia ja kehotetta. Sitten tulostamme tuotetun tekstin.

### Chat-vastaukset

Olet nähnyt siihen asti miten voimme tuottaa tekstiä yhdestä kehotteesta. Responses-API on myös hyvin soveltuva chatbotteihin – annat viestilistan `input`-parametrina rakentamaan keskustelun. Tässä esimerkki sen käyttämisestä:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Lisätietoa tästä toiminnallisuudesta tulee tulevassa luvussa.

## Harjoitus - rakenna reseptigeneraattori

Nyt kun olemme pelanneet skenaarion läpi, kirjoitetaan koodi vastaamaan esitettyä skenaariota. Toimi seuraavasti:

1. Käytä lähtökohtana olemassa olevaa *app.py*-tiedostoa
1. Etsi muuttuja `prompt` ja vaihda sen koodi seuraavanlaiseksi:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> HUOMAUTUS, LLM:si on epädeterministinen, joten saatat saada erilaisia tuloksia joka kerta kun suoritat ohjelman.

Hienoa, katsotaan miten voimme parantaa asioita. Parantaaksemme asioita, haluamme varmistaa että koodi on joustavaa, jotta ainesosia ja reseptien määrä voidaan parantaa ja muuttaa. 


1. Muutetaan koodia seuraavalla tavalla:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Tässä ehdotamme ylimääräisen kehotteen lisäämistä, mutta sen toimimiseksi meidän on lisättävä aiemman kehotteen tulos osaksi kontekstia myöhemmälle kehotteelle. 

    Etsi koodista kohta, joka tulostaa ensimmäisen kehotteen tuloksen, ja lisää seuraava koodi sen alle:

    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # tulosta vastaus
    print("Shopping list:")
    print(response.output_text)
    ```

- **Sana token-pituudesta**. Meidän tulisi harkita, kuinka monta tokenia tarvitsemme halutun tekstin tuottamiseen. Tokenit maksavat, joten missä mahdollista, meidän tulisi olla säästäväisiä käyttämissämme tokenien määrässä. Esimerkiksi, voimmeko muotoilla kehotteen niin, että voimme käyttää vähemmän tokeneita?

   Tokenien määrää voi muuttaa `max_output_tokens` -parametrilla. Esimerkiksi, jos haluat käyttää 100 tokenia, tekisit näin:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Kokeile lämpötilaa**. Lämpötila on asia, jota emme ole tähän mennessä maininneet, mutta se on tärkeä tekijä siinä, miten ohjelmamme toimii. Mitä korkeampi lämpötila-arvo, sitä satunnaisempi tuloste on. Päinvastoin, mitä matalampi lämpötila-arvo, sitä ennustettavampi tuloste on. Mieti, haluatko vaihtelua tulosteeseesi vai et.

   Lämpötilaa voi muuttaa parametrilla `temperature`. Esimerkiksi, jos haluat käyttää lämpötilaa 0.5, tekisit näin:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Huomaa, mitä lähempänä arvo on 1.0, sitä vaihtelevampi on tuloste.
