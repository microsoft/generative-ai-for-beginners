## Mikä on tekstintuotantosovellus?

Tavallisesti, kun rakennat sovelluksen, sillä on jonkinlainen käyttöliittymä kuten seuraava:

- Komentorivipohjainen. Konsolisovellukset ovat tyypillisiä sovelluksia, joissa kirjoitat komennon ja se suorittaa tehtävän. Esimerkiksi `git` on komentopohjainen sovellus.
- Käyttöliittymä (UI). Joissain sovelluksissa on graafiset käyttöliittymät (GUI), joissa klikkaat painikkeita, syötät tekstiä, valitset vaihtoehtoja ja muuta.

### Komentorivi- ja käyttöliittymäsovellukset ovat rajoitettuja

Vertaa sitä komentopohjaiseen sovellukseen, jossa kirjoitat komennon:

- **Se on rajoitettua**. Et voi kirjoittaa mitä tahansa komentoa, vain niitä joita sovellus tukee.
- **Kielikohtainen**. Jotkut sovellukset tukevat monia kieliä, mutta oletuksena sovellus on rakennettu tietylle kielelle, vaikka voit lisätä muutakin kielitukea.

### Tekstintuotantosovellusten hyödyt

Mikä sitten tekee tekstintuotantosovelluksesta erilaisen?

Tekstintuotantosovelluksessa sinulla on enemmän joustavuutta, et ole sidottu joukkoon komentoja tai tiettyyn syötekieleen. Sen sijaan voit käyttää luonnollista kieltä vuorovaikutukseen sovelluksen kanssa. Toinen hyöty on, että koska olet jo vuorovaikutuksessa tietolähteen kanssa, joka on koulutettu valtavalla tietomäärällä, perinteinen sovellus voi olla rajoitettu siihen, mitä tietokannassa on.

### Mitä voin rakentaa tekstintuotantosovelluksella?

Monet asiat ovat mahdollisia. Esimerkiksi:

- **Chatbotin**. Chatbot, joka vastaa kysymyksiin aiheista, kuten yrityksestäsi ja sen tuotteista, voisi olla hyvä sovellus.
- **Avustajan**. LLM:t ovat loistavia tekstin tiivistämisessä, tekstistä oivallusten saamisessa, tekstin tuottamisessa kuten ansioluetteloissa ja muussa.
- **Koodiavustajan**. Riippuen käyttämästäsi kielimallista, voit rakentaa koodiavustajan, joka auttaa kirjoittamaan koodia. Esimerkiksi voit käyttää GitHub Copilot -tuotetta sekä ChatGPT:tä auttamaan koodin kirjoittamisessa.

## Generoi tekstiä

Tapa generoida tekstiä on käyttää `responses.create`-metodia. Tässä esimerkki:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yllä olevassa koodissa luomme vastauksen ja annamme mallin, jota haluamme käyttää, sekä kehotteen. Sitten tulostamme generoitu tekstin.

### Chat-vastaukset

Tähän asti olet nähnyt miten voimme generoida tekstiä yhdestä kehotteesta. Responses API sopii myös hyvin chatboteille - annat viestilistan `input`:na keskustelun rakentamiseksi. Tässä esimerkki sen käytöstä:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Lisää tästä toiminnallisuudesta seuraavassa luvussa.

Nyt kun olemme käyneet läpi skenaarion, kirjoitetaan koodi vastaamaan demonstroitua skenaariota. Toimi seuraavasti:

1. Käytä olemassa olevaa *app.py*-tiedostoa lähtökohtana
1. Etsi `prompt`-muuttuja ja muuta sen koodi seuraavanlaiseksi:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> HUOMAUTUS, LLM:si on epädeterministinen, joten saatat saada eri tuloksia joka kerta kun ajat ohjelman.

Hienoa, katsotaan kuinka voimme parantaa asioita. Parantaaksemme asioita, haluamme varmistaa, että koodi on joustava, jotta aineksia ja reseptien määrää voidaan parantaa ja muuttaa.


1. Muutetaan koodia seuraavalla tavalla:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Paranna lisäämällä suodatin ja ostoslista

Meillä on nyt toimiva sovellus, joka pystyy tuottamaan reseptejä, ja se on joustava, koska se perustuu käyttäjän syötteisiin, sekä reseptien määrään että käytettyihin aineksiin.

    Tätä toimintoa varten voimme yrittää ratkaista kaiken yhdellä kehotteella tai jakaa sen kahteen kehotteeseen. Kokeillaan jälkimmäistä lähestymistapaa. Tässä ehdotamme lisäkehotteen lisäämistä, mutta tämän toimimiseksi meidän on lisättävä ensimmäisen kehotteen tulos kontekstiksi toiseen kehotteeseen. 

    Etsi koodista kohta, joka tulostaa ensimmäisen kehotteen tuloksen, ja lisää alle seuraava koodi:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # tulosta vastaus
    print("Shopping list:")
    print(response.output_text)
    ```

    Huomaa seuraavat asiat:

    - Rakennamme uuden kehotteen lisäämällä ensimmäisen kehotteen tuloksen uuteen kehotteeseen: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Teemme uuden pyynnön, mutta otamme myös huomioon ensimmäisessä kehotteessa pyytämämme token-määrän, joten tällä kertaa asetamme `max_output_tokens` arvoksi 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

- **Sana token-pituudesta**. Meidän tulisi huomioida, kuinka monta tokenia tarvitsemme halutun tekstin tuottamiseen. Tokenit maksavat rahaa, joten missä mahdollista meidän tulisi olla taloudellisia käyttämiemme tokenien määrän kanssa. Voisimmeko esimerkiksi muotoilla kehottetta niin, että voisimme käyttää vähemmän tokeneita?

   Voit muuttaa käytettyjen tokenien määrää käyttämällä parametria `max_output_tokens`. Esimerkiksi, jos haluat käyttää 100 tokenia, teet näin:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Kokeile lämpötilaa**. Lämpötila on jotain, mitä emme ole maininneet aiemmin, mutta se on tärkeä konteksti sille, miten ohjelmamme toimii. Mitä korkeampi lämpötila-arvo on, sitä satunnaisempi on tuloste. Mitä matalampi lämpötila-arvo on, sitä ennustettavampi on tuloste. Mieti, haluatko vaihtelua tulosteeseesi vai et.

   Muuttaaksesi lämpötilaa voit käyttää parametria `temperature`. Esimerkiksi, jos haluat käyttää lämpötilaa 0.5, teet näin:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Huomaa, mitä lähempänä arvo on 1.0, sitä vaihtelevampi on tuloste.
