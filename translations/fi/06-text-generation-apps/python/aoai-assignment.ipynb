## Mikä on tekstin generointisovellus?

Normaali sovellus sisältää jonkinlaisen käyttöliittymän, kuten seuraavasti:

- Komentorivipohjainen. Konsolisovellukset ovat tyypillisiä, joissa kirjoitat komennon ja se suorittaa tehtävän. Esimerkiksi `git` on komentopohjainen sovellus.
- Käyttöliittymä (UI). Joissain sovelluksissa on graafinen käyttöliittymä (GUI), jossa klikkaat painikkeita, syötät tekstiä, valitset vaihtoehtoja ja muuta.

### Konsoli- ja käyttöliittymäsovellukset ovat rajallisia

Vertaa tätä komentopohjaiseen sovellukseen, jossa kirjoitat komennon: 

- **Se on rajallinen**. Et voi vain kirjoittaa mitä tahansa komentoa, vaan vain sellaisia, joita sovellus tukee.
- **Kielikohtainen**. Jotkin sovellukset tukevat useita kieliä, mutta oletuksena sovellus on rakennettu tietylle kielelle, vaikka voit lisätä enemmän kielitukea.

### Tekstin generointisovellusten edut

Mikä sitten tekee tekstin generointisovelluksesta erilaisen?

Tekstin generointisovelluksessa sinulla on enemmän joustavuutta, et ole rajoitettu joukkoon komentoja tai tiettyyn syöttökieleen. Sen sijaan voit käyttää luonnollista kieltä kommunikoimiseen sovelluksen kanssa. Toinen etu on, että koska olet vuorovaikutuksessa tietolähteen kanssa, joka on koulutettu laajan tietoaineiston perusteella, perinteinen sovellus saattaa olla rajallinen siinä, mitä tietokannasta löytyy.

### Mitä voin rakentaa tekstin generointisovelluksella?

Voit rakentaa monenlaisia asioita. Esimerkiksi:

- **Chatbotin**. Chatbot, joka vastaa kysymyksiin aiheista kuten yrityksestäsi ja sen tuotteista, voisi olla hyvä ratkaisu.
- **Avustajan**. Suuret kielimallit (LLM) ovat erinomaisia esimerkiksi tekstin tiivistämisessä, tekstistä saatavien havaintojen tekemisessä, tekstin tuottamisessa kuten ansioluetteloiden kirjoittamisessa ja muussa.
- **Koodiavustajan**. Riippuen käyttämästäsi kielimallista, voit rakentaa koodiavustajan, joka auttaa sinua kirjoittamaan koodia. Esimerkiksi voit käyttää tuotteita kuten GitHub Copilot sekä ChatGPT auttamaan koodin kirjoittamisessa.


Yllä asetamme seuraavaa:

- `deployment`, tämä on mallin käyttöönoton nimi, jonka loit Foundry-portaalissa.

## Generoi tekstiä

Tapa tuottaa tekstiä on käyttää `responses.create` -metodia. Tässä esimerkki:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Edellä olevassa koodissa luomme vastausobjektin ja annamme käyttöön halutun mallin ja kehotteen. Sitten tulostamme generoidun tekstin.

### Chat-vastaukset

Responses API sopii hyvin sekä yksittäisen vuoron tekstin generointiin että monivuoroisiin chatbotteihin – tarjoat vain lisää viestejä `input`-listassa rakentamaan keskustelun:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],

```

Tätä toimintoa käsitellään tarkemmin tulevassa luvussa.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

    ```python
    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
