## Mikä on tekstin generointisovellus?

Tavallisesti, kun rakennat sovelluksen, siellä on jokin käyttöliittymä, kuten seuraava:

- Komentopohjainen. Konsolisovellukset ovat tyypillisiä sovelluksia, joissa kirjoitat komennon ja se suorittaa tehtävän. Esimerkiksi `git` on komentopohjainen sovellus.
- Käyttöliittymä (UI). Joissakin sovelluksissa on graafinen käyttöliittymä (GUI), jossa painat nappeja, syötät tekstiä, valitset vaihtoehtoja ja muuta.

### Konsoli- ja käyttöliittymäsovellukset ovat rajallisia

Vertaa sitä komentopohjaiseen sovellukseen, jossa kirjoitat komennon: 

- **Se on rajallista**. Et voi kirjoittaa mitä tahansa komentoa, vaan vain niitä, joita sovellus tukee.
- **Kielikohtainen**. Jotkut sovellukset tukevat monia kieliä, mutta oletuksena sovellus on rakennettu tietylle kielelle, vaikka voitkin lisätä kielitukea.

### Tekstin generointisovellusten edut

Mikä sitten erottaa tekstin generointisovelluksen?

Tekstin generointisovelluksessa sinulla on enemmän joustavuutta, et ole rajattu tiettyihin komentoihin tai tiettyyn syötekieleen. Sen sijaan voit käyttää luonnollista kieltä olla vuorovaikutuksessa sovelluksen kanssa. Toinen etu on, että koska olet jo vuorovaikutuksessa tietolähteen kanssa, joka on koulutettu laajalla informaation kokoelmalla, perinteinen sovellus saattaa olla rajattu vain tietokannan sisältöön.

### Mitä voin rakentaa tekstin generointisovelluksella?

Voit rakentaa monia asioita. Esimerkiksi:

- **Chatbotin**. Chatbot, joka vastaa kysymyksiin aiheista, kuten yrityksestäsi ja sen tuotteista, voisi olla hyvä valinta.
- **Avustajan**. Suuret kielimallit (LLM) ovat erinomaisia tehtävissä kuten tekstin tiivistäminen, tekstin tulosten saanti, tekstin tuottaminen kuten ansioluettelot ja muuta.
- **Koodiavustajan**. Käyttämästäsi kielimallista riippuen voit rakentaa koodiavustajan, joka auttaa sinua kirjoittamaan koodia. Esimerkiksi voit käyttää tuotteita kuten GitHub Copilot sekä ChatGPT:n auttamaan sinua koodin kirjoittamisessa.

## Tekstin generointi

Tekstin generoimiseen käytetään `responses.create` -metodia. Tässä esimerkki:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yllä olevassa koodissa luomme vastauksen objektin ja annamme mallin, jota haluamme käyttää, sekä kehotteen. Sitten tulostamme generoidun tekstin.

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```
