# मजकूर निर्मिती अ‍ॅप्स तयार करा

तुम्ही आता पर्यंत या अभ्यासक्रमामध्ये पाहिले आहे की, प्रॉम्प्टसारख्या मूलभूत संकल्पना आहेत आणि "प्रॉम्प्ट अभियांत्रण" नावाने एक संपूर्ण शिस्तही आहे. ChatGPT, Office 365, Microsoft Power Platform आणि अनेक इतर उपकरणांमध्ये तुम्ही प्रॉम्प्ट्स वापरून काही साध्य करू शकता.

अशा अनुभवाला अ‍ॅपमध्ये जोडण्यासाठी, तुम्हाला प्रॉम्प्ट्स, पूर्णता यांसारख्या संकल्पना समजून घ्याव्या लागतील आणि काम करण्यासाठी एक लायब्ररी निवडावी लागेल. हेच तुम्ही या प्रकरणात शिकणार आहात.

## परिचय

या प्रकरणात, तुम्ही:

- openai लायब्ररी आणि तिच्या मूलभूत संकल्पना शिकाल.
- openai वापरून मजकूर निर्मिती अ‍ॅप तयार कराल.
- प्रॉम्प्ट, तापमान आणि टोकन्स यांसारख्या संकल्पनांचा वापर करून मजकूर निर्मिती अ‍ॅप कसे तयार करायचे ते समजल.

## शिक्षण उद्दिष्टे

या धड्याच्या शेवटी, तुम्ही सक्षम असाल:

- मजकूर निर्मिती अ‍ॅप काय आहे हे समजावून सांगायला.
- openai वापरून मजकूर निर्मिती अ‍ॅप तयार करायला.
- अधिक किंवा कमी टोकन्स वापरण्यासाठी आणि परिणाम वेगळ्या स्वरूपात मिळवण्यासाठी तापमान कसे बदलायचे ते तुमच्या अ‍ॅपमध्ये सेटअप करायला.

## मजकूर निर्मिती अ‍ॅप काय आहे?

सामान्यतः जेव्हा तुम्ही अ‍ॅप तयार करता त्यात खालीलप्रमाणे काही प्रकारचे इंटरफेस असते:

- आदेश-आधारित. कन्सोल अ‍ॅप्स हे असे अ‍ॅप्स असतात जिथे तुम्ही आदेश टाइप करता आणि ते कार्य पार पाडते. उदाहरणार्थ, `git` हा आदेश-आधारित अ‍ॅप आहे.
- वापरकर्ता इंटरफेस (UI). काही अ‍ॅप्समध्ये ग्राफिकल वापरकर्ता इंटरफेस (GUI) असतो जिथे तुम्ही बटणे क्लिक करता, मजकूर घालता, पर्याय निवडता आणि बरेच काही करता.

### कन्सोल आणि UI अ‍ॅप्सची मर्यादा

त्याची तुलना करा अशा आदेश-आधारित अ‍ॅपशी जिथे तुम्ही आदेश टाइप करता: 

- **हे मर्यादित आहे**. तुम्ही कोणतेही आदेश टाइप करू शकत नाही, फक्त अ‍ॅपने समर्थित आदेश वापरू शकता.
- **भाषा विशिष्ट**. काही अ‍ॅप्स अनेक भाषांचा समर्थन करतात, पण मूलत: अ‍ॅप विशिष्ट भाषेसाठी तयार केलेला असतो, जरी तुम्ही अधिक भाषा समर्थन जोडू शकता.

### मजकूर निर्मिती अ‍ॅप्सच्या फायदे

चांगले तर मजकूर निर्मिती अ‍ॅप कसा वेगळा आहे?

मजकूर निर्मिती अ‍ॅपमध्ये, तुम्हाला अधिक लवचिकता असते, तुम्ही आदेशांच्या सेट किंवा विशिष्ट इनपुट भाषेपुरतं मर्यादित नसता. त्याऐवजी, तुम्ही नैसर्गिक भाषा वापरून अ‍ॅपशी संवाद साधू शकता. आणखी एक फायदा म्हणजे तुम्ही आधीच मोठ्या प्रमाणातील माहितीवर प्रशिक्षित केलेल्या डेटास्रोताशी संवाद साधत आहात, ज्यामध्ये पारंपरिक अ‍ॅप डेटाबेसच्या मर्यादेपेक्षा जास्त माहिती आहे.

### मजकूर निर्मिती अ‍ॅप वापरून मी काय तयार करू शकतो?

अनेक गोष्टी तयार करता येतील. उदाहरणार्थ:

- **एक चॅटबॉट**. तुमच्या कंपनी आणि तिच्या उत्पादनांबद्दल प्रश्नांची उत्तरे देणारा चॅटबॉट चांगला पर्याय असू शकतो.
- **सहायक**. LLMs मजकूराचे सारांश बनवणे, मजकूरातून अंतर्दृष्टी मिळवणे, रेझ्युमे सारखा मजकूर तयार करणे यांसारख्या गोष्टींच्या दृष्टीने उत्कृष्ट आहेत.
- **कोड सहाय्यक**. तुम्ही वापरलेल्या भाषा मॉडेलवर अवलंबून, तुम्ही कोड लिहिण्यास मदत करणारा कोड सहाय्यक तयार करू शकता. उदाहरणार्थ, GitHub Copilot किंवा ChatGPT सारख्या उत्पादनांचा वापर करून कोड लिहिण्याची मदत घेऊ शकता.

## मी कसा सुरुवात करू शकतो?

बरं, तुम्हाला LLM सह एकत्रीकरण करण्याचा मार्ग शोधावा लागेल, जो सामान्यतः खालील दोन पद्धतींचा समावेश करतो:

- API वापरा. येथे तुम्ही तुमचा प्रॉम्प्टसह वेब विनंत्या तयार करता आणि तयार केलेला मजकूर परत मिळवता.
- लायब्ररी वापरा. लायब्ररी API कॉल्समध्ये समाविष्ट करून त्याचा वापर सुलभ करतात.

## लायब्ररीज/SDKs

LLMs सोबत काम करण्यासाठी काही प्रसिद्ध लायब्ररीज आहेत जसे की:

- **openai**, ही लायब्ररी तुमच्या मॉडेलशी जोडणे आणि प्रॉम्प्ट्स पाठविणे सोपे करते.

त्यानंतर उच्चस्तरीय लायब्ररीज देखील आहेत जसे की:

- **Langchain**. Langchain प्रसिद्ध आहे आणि Python ला समर्थन करते.
- **Semantic Kernel**. Semantic Kernel हे Microsoft कडून एक लायब्ररी आहे जी C#, Python, आणि Java भाषांना समर्थन देते.

## openai वापरून प्रथम अ‍ॅप

आता बघूया की आपण आपला पहिला अ‍ॅप कसा तयार करू, कोणत्या लायब्ररींची गरज आहे, किती गरजेचे आहे आणि पुढे काय.

### openai ची स्थापना करा

आम्ही `openai` Python लायब्ररी वापरण्याचा निर्णय घेतला आहे, म्हणून `pip` वापरून ती स्थापित करू.

### एक संसाधन तयार करा आणि तुमचा API की शोधा

जर तुम्ही आधीच केले नसेल, तर खालील पायऱ्या पार पाडाव्यात:

- OpenAI वर खाते तयार करा <https://platform.openai.com/signup>.
- आता तुमचे API की मिळवा <https://platform.openai.com/api-keys>.

>[!NOTE]
> तुमचा API की तुमच्या कोडपासून वेगळा ठेवणे उपयुक्त आहे. तुम्ही वातावरणातील चल (environment variables) वापरून तसे करू शकता.
> - तुमच्या .env फाईलमध्ये `OPENAI_KEY` या environment variable मध्ये तुमचा API की सेट करा. जर तुम्ही आधीच या कोर्सच्या मागील अभ्यास पूर्ण केले असतील तर तुम्ही तयार आहात.
> - लक्षात ठेवा की API की फक्त एकदाच उपलब्ध होते. त्यामुळे त्याची कॉपी बरोबर झाली आहे की नाही हे तपासणे महत्त्वाचे आहे. जर ते काम केले नाही तर की डिलीट करून नवीन की तयार करण्याचा सल्ला दिला जातो.


### OpenAI कॉन्फिगरेशन सेटअप करा

OpenAI वापरत असाल तर कॉन्फिगरेशन सेटअप करण्याचा प्रकार खालीलप्रमाणे आहे:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

वर आपण खालील सेट करत आहोत:

- `api_key`, हा तुमचा OpenAI डॅशबोर्डमध्ये आढळणारा API की आहे.
- `deployment`, ही तुमची GPT आवृत्ती आहे.

> [!NOTE]
> `os.environ` ही एक फंक्शन आहे जी environment variables वाचते. तुम्ही याचा वापर `OPENAI_API_KEY` सारखे environment variables वाचण्यासाठी करू शकता.

## मजकूर तयार करा

मजकूर तयार करण्याचा मार्ग म्हणजे `responses.create` पद्धत वापरणे. उदाहरण खाली दिले आहे:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

वर दिलेल्या कोडमध्ये, आपण एक response ऑब्जेक्ट तयार करतो आणि वापरायचं मॉडेल आणि प्रॉम्प्ट पास करतो. नंतर आपण तयार केलेला मजकूर प्रिंट करतो.

### चॅट प्रतिसाद

आतापर्यंत, तुम्ही पाहिले की आपण एकाच प्रॉम्प्टवरून मजकूर कसा तयार करू शकतो. Responses API चॅटबॉटसाठीही उपयुक्त आहे - तुम्ही `input` मध्ये संदेशांची यादी द्याल ज्याने संभाषण तयार होईल. त्याचा वापर कसा करायचा हे खाली दिले आहे:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

या कार्यक्षमतेबद्दल पुढील प्रकरणात अधिक सविस्तर माहिती दिली जाईल.

## व्यायाम - तुमचा पहिला मजकूर निर्मिती अ‍ॅप

आता आपण OpenAI सेवा कसे सेटअप आणि कॉन्फिगर करायचे ते शिकलो आहे, आता तुमचा पहिला मजकूर निर्मिती अ‍ॅप तयार करण्याची वेळ आली आहे. तुमचा अ‍ॅप तयार करण्यासाठी खालील टप्यांचे अनुसरण करा:


1. एक व्हर्चुअल एन्व्हायर्नमेंट तयार करा आणि openai इन्स्टॉल करा:

  > [!NOTE] जर तुम्ही हा नोटबुक Codespaces वर किंवा Devcontainer मध्ये चालवत असाल तर हा टप्पा आवश्यक नाही


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> आपल्याला Windows वापरत असल्यास `source venv/bin/activate` ऐवजी `venv\Scripts\activate` टाईप करा.

> [!NOTE]
> आपली OpenAI की शोधण्यासाठी https://platform.openai.com/settings/organization/api-keys वर जा आणि `API keys` शोधा. आपण तिथे नवीन की तयार करू शकता आणि लगेच त्याचा मूल्य कॉपी करू शकता.


1. *app.py* फाइल तयार करा आणि त्यात खालील कोड द्या:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    तुम्हाला खालीलप्रमाणे आउटपुट दिसायला पाहिजे:

    ```output
     खूप दुःखी _____.

    एकदा काळी एक खूप दुःखी जलपरि होती.
    ```


## वेगवेगळ्या प्रकारच्या प्रॉम्प्ट्स, वेगवेगळ्या गोष्टींसाठी

तुम्ही आता पाहिले आहे की प्रॉम्प्ट वापरून कसा मजकूर तयार करायचा. तुमच्याकडे एक प्रोग्राम देखील चालू आहे ज्यामध्ये तुम्ही बदल करून वेगवेगळ्या प्रकारच्या मजकुराची निर्मिती करू शकता.

प्रॉम्प्ट्स विविध प्रकारच्या कामांसाठी वापरले जाऊ शकतात. उदाहरणार्थ:

- **मजकुराचा एखादा प्रकार तयार करा**. उदाहरणार्थ, तुम्ही कविता, प्रश्नांची एक परीक्षा यांसारखे मजकूर तयार करू शकता.
- **माहिती शोधा**. तुम्ही प्रॉम्प्ट्सचा वापर करून खालील उदाहरणाप्रमाणे माहिती शोधू शकता 'वेब डेव्हलपमेंटमध्ये CORS म्हणजे काय?'.
- **कोड तयार करा**. तुम्ही प्रॉम्प्ट्सचा वापर करून कोड तयार करू शकता, उदाहरणार्थ ईमेल्सची पडताळणी करण्यासाठी रेग्युलर एक्सप्रेशन तयार करणे किंवा पूर्ण प्रोग्रामसारखे वेब अॅप तयार करणे.

## एक अधिक व्यावहारिक वापर: रेसिपी जनरेटर

समजा तुमच्याकडे घरच्या घरी साहित्य आहे आणि तुम्हाला काहीतरी स्वयंपाक करायचा आहे. त्यासाठी तुला रेसिपीची गरज आहे. रेसिपी शोधण्याचा एक मार्ग म्हणजे सर्च इंजिन वापरणे किंवा तुम्ही LLM चा उपयोग करू शकता.

तुम्ही असा प्रॉम्प्ट लिहू शकता:

1. Roasted Chicken and Vegetables:

- 2 tablespoons olive oil
- 2 cloves garlic, minced
- 1 teaspoon dried thyme

आता आपल्याने एक परिस्थिती सादर केली आहे, चला दाखवलेल्या परिस्थितीशी जुळणारा कोड लिहूया. तसे करण्यासाठी, खालील चरणांचे पालन करा:

1. सुरू करण्यासाठी विद्यमान *app.py* फाइल वापरा
1. `prompt` व्हेरिएबल शोधा आणि त्याचा कोड खालीलप्रमाणे बदला:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. हे आउटपुट किती अनियमित आहे हे नियंत्रित करते.
1. हे प्रतिसाद किती मोठा आहे हे नियंत्रित करते.
1. हे किती टोकन्स वापरले जातात हे नियंत्रित करते.

उत्तर: 1

API कीसारखे रहस्य कसे चांगल्या प्रकारे जतन करावे?

1. कोडमध्ये.
1. फाइलमध्ये.
1. पर्यावरणीय चल (environment variables) मध्ये.

उत्तर: 3, कारण पर्यावरणीय चल कोडमध्ये जतन केले जात नाहीत आणि कोडमधून लोड केले जाऊ शकतात.
