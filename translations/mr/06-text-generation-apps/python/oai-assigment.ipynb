# टेक्स्ट जनरेशन अ‍ॅप्स तयार करा

या अभ्यासक्रमात तुम्ही आतापर्यंत पाहिलं आहे की प्रॉम्प्ट्ससारख्या मुख्य संकल्पना आहेत आणि "प्रॉम्प्ट इंजिनिअरिंग" नावाचं एक संपूर्ण क्षेत्र आहे. ChatGPT, Office 365, Microsoft Power Platform आणि इतर अनेक टूल्समध्ये तुम्ही प्रॉम्प्ट्स वापरून काहीतरी साध्य करू शकता.

तुमच्या अ‍ॅपमध्ये अशी सुविधा जोडायची असेल, तर तुम्हाला प्रॉम्प्ट्स, कम्प्लिशन्स यांसारख्या संकल्पना समजून घ्याव्या लागतील आणि कोणती लायब्ररी वापरायची हे ठरवावं लागेल. ह्याच गोष्टी तुम्ही या अध्यायात शिकणार आहात.

## परिचय

या अध्यायात, तुम्ही:

- openai लायब्ररी आणि तिच्या मुख्य संकल्पना शिकाल.
- openai वापरून टेक्स्ट जनरेशन अ‍ॅप तयार कराल.
- प्रॉम्प्ट, टेम्परेचर, आणि टोकन्ससारख्या संकल्पना वापरून टेक्स्ट जनरेशन अ‍ॅप कसं तयार करायचं हे समजून घ्याल.

## शिकण्याची उद्दिष्टे

या धड्याच्या शेवटी, तुम्ही हे करू शकाल:

- टेक्स्ट जनरेशन अ‍ॅप म्हणजे काय हे समजावून सांगू शकाल.
- openai वापरून टेक्स्ट जनरेशन अ‍ॅप तयार करू शकाल.
- तुमच्या अ‍ॅपमध्ये जास्त किंवा कमी टोकन्स वापरण्यासाठी आणि वेगळ्या आउटपुटसाठी टेम्परेचर बदलण्यासाठी सेटिंग्ज करू शकाल.

## टेक्स्ट जनरेशन अ‍ॅप म्हणजे काय?

साधारणपणे जेव्हा तुम्ही अ‍ॅप तयार करता, तेव्हा त्याला खालीलपैकी काही इंटरफेस असतो:

- कमांड-आधारित. कन्सोल अ‍ॅप्स म्हणजे जिथे तुम्ही कमांड टाकता आणि ते काम करतात. उदाहरणार्थ, `git` हे कमांड-आधारित अ‍ॅप आहे.
- युजर इंटरफेस (UI). काही अ‍ॅप्सना ग्राफिकल युजर इंटरफेस (GUI) असतो जिथे तुम्ही बटणं क्लिक करता, मजकूर टाकता, पर्याय निवडता इ.

### कन्सोल आणि UI अ‍ॅप्स मर्यादित असतात

कमांड-आधारित अ‍ॅपशी तुलना करा:

- **मर्यादित आहे**. तुम्ही कोणतीही कमांड टाकू शकत नाही, फक्त त्या अ‍ॅपने सपोर्ट केलेल्या कमांड्सच वापरू शकता.
- **भाषेसाठी खास**. काही अ‍ॅप्स अनेक भाषा सपोर्ट करतात, पण बहुतेक वेळा अ‍ॅप विशिष्ट भाषेसाठीच तयार केलेलं असतं, जरी तुम्ही इतर भाषा जोडू शकत असलात तरी.

### टेक्स्ट जनरेशन अ‍ॅप्सचे फायदे

मग टेक्स्ट जनरेशन अ‍ॅप वेगळं कसं आहे?

टेक्स्ट जनरेशन अ‍ॅपमध्ये तुम्हाला जास्त लवचिकता असते, तुम्ही फक्त ठराविक कमांड्स किंवा इनपुट भाषेपुरते मर्यादित राहत नाही. त्याऐवजी, तुम्ही नैसर्गिक भाषेतून अ‍ॅपशी संवाद साधू शकता. आणखी एक फायदा म्हणजे तुम्ही अशा डेटासोर्सशी संवाद साधत आहात ज्यावर प्रचंड माहितीवर ट्रेनिंग झालं आहे, तर पारंपारिक अ‍ॅप फक्त डेटाबेसपुरताच मर्यादित असू शकतो.

### टेक्स्ट जनरेशन अ‍ॅप वापरून मी काय तयार करू शकतो?

तुम्ही अनेक गोष्टी तयार करू शकता. उदाहरणार्थ:

- **चॅटबॉट**. एखाद्या विषयावर प्रश्नांची उत्तरे देणारा चॅटबॉट, जसं की तुमची कंपनी आणि तिची उत्पादने, हे चांगलं उदाहरण आहे.
- **सहाय्यक**. LLMs मजकूर संक्षिप्त करणे, मजकुरातून माहिती मिळवणे, रिझ्युमेसारखा मजकूर तयार करणे इ. गोष्टींमध्ये खूप चांगले आहेत.
- **कोड असिस्टंट**. तुम्ही कोणता भाषा मॉडेल वापरता यावर अवलंबून, तुम्ही कोड लिहायला मदत करणारा कोड असिस्टंट तयार करू शकता. उदाहरणार्थ, GitHub Copilot किंवा ChatGPT वापरून तुम्ही कोड लिहायला मदत मिळवू शकता.

## मी कसा सुरुवात करू शकतो?

तुम्हाला LLM सोबत इंटिग्रेट करायचं असेल, तर साधारणपणे दोन मार्ग असतात:

- API वापरा. इथे तुम्ही प्रॉम्प्टसह वेब रिक्वेस्ट तयार करता आणि त्यावरून जनरेट केलेला मजकूर मिळवता.
- लायब्ररी वापरा. लायब्ररी API कॉल्स सहज वापरता येतील अशा पद्धतीने तयार करतात.

## लायब्ररी/SDKs

LLMs सोबत काम करण्यासाठी काही प्रसिद्ध लायब्ररी आहेत:

- **openai**, ही लायब्ररी तुमच्या मॉडेलशी कनेक्ट होऊन प्रॉम्प्ट्स पाठवणं सोपं करते.

त्यानंतर काही लायब्ररी आहेत ज्या उच्च पातळीवर काम करतात:

- **Langchain**. Langchain प्रसिद्ध आहे आणि Python ला सपोर्ट करते.
- **Semantic Kernel**. Semantic Kernel ही Microsoft ची लायब्ररी आहे जी C#, Python, आणि Java या भाषांना सपोर्ट करते.

## openai वापरून पहिलं अ‍ॅप

चला पाहूया की आपलं पहिलं अ‍ॅप कसं तयार करायचं, कोणत्या लायब्ररी लागतील, किती कोड लागेल इ.

### openai इन्स्टॉल करा

  > [!NOTE] जर तुम्ही हे नोटबुक Codespaces किंवा Devcontainer मध्ये चालवत असाल तर हा टप्पा आवश्यक नाही


OpenAI किंवा Azure OpenAI सोबत संवाद साधण्यासाठी अनेक लायब्ररी उपलब्ध आहेत. C#, Python, JavaScript, Java आणि इतर अनेक प्रोग्रामिंग भाषा वापरता येतात.  
आम्ही `openai` Python लायब्ररी वापरण्याचा निर्णय घेतला आहे, म्हणून `pip` वापरून ती इन्स्टॉल करू.

```bash
pip install openai
```

जर तुम्ही हे नोटबुक Codespaces किंवा Dev Container मध्ये चालवत नसाल, तर [Python](https://www.python.org/) तुमच्या मशीनवर इन्स्टॉल करावं लागेल.

### रिसोर्स तयार करा आणि तुमची API key शोधा

जर तुम्ही आधीच केलं नसेल, तर खालील स्टेप्स फॉलो करा:

- OpenAI वर अकाउंट तयार करा <https://platform.openai.com/signup>.
- आता, तुमची API keys मिळवा <https://platform.openai.com/api-keys>. 

>[!NOTE]
> तुमची API key कोडपासून वेगळी ठेवणं चांगलं असतं. हे environment variables वापरून करता येतं.
> - .env फाइलमध्ये `OPENAI_KEY` environment variable म्हणून तुमची API key सेट करा. जर तुम्ही या कोर्समधील मागील सराव पूर्ण केला असेल, तर हे आधीच सेट झालं असेल.
> - लक्षात ठेवा की API Key फक्त एकदाच दिसते. त्यामुळे ती बरोबर कॉपी झाली आहे का हे तपासा. जर ती काम करत नसेल, तर ती डिलीट करून नवीन key जनरेट करा.


### OpenAI ची कॉन्फिगरेशन सेटअप करा

जर तुम्ही OpenAI वापरत असाल, तर कॉन्फिगरेशन कसं सेट करायचं:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

वर आपण हे सेट करत आहोत:

- `api_key`, ही तुमची OpenAI डॅशबोर्डवरून मिळालेली API key आहे.
- `deployment`, हा तुमचा GPT व्हर्जन आहे.

> [!NOTE]
> `os.environ` ही फंक्शन environment variables वाचते. तुम्ही `OPENAI_API_KEY` सारखे environment variables वाचण्यासाठी हे वापरू शकता.

## टेक्स्ट जनरेट करा

टेक्स्ट जनरेट करण्याचा मार्ग म्हणजे `chat.completion` क्लास वापरणं. उदाहरण:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

वरील कोडमध्ये, आपण एक completion object तयार करतो आणि वापरायचा मॉडेल व प्रॉम्प्ट देतो. मग आपण जनरेट केलेला मजकूर प्रिंट करतो.

### Chat completions

आत्तापर्यंत आपण `Completion` वापरून टेक्स्ट जनरेट करत होतो. पण अजून एक क्लास आहे `ChatCompletion` जो चॅटबॉटसाठी जास्त योग्य आहे. हे कसं वापरायचं त्याचं उदाहरण:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

या फिचरबद्दल पुढच्या अध्यायात अधिक माहिती मिळेल.

## सराव - तुमचं पहिलं टेक्स्ट जनरेशन अ‍ॅप

आता आपण OpenAI सेवा कशी सेटअप करायची आणि कॉन्फिगर करायची हे शिकलो, आता तुमचं पहिलं टेक्स्ट जनरेशन अ‍ॅप तयार करूया. अ‍ॅप तयार करण्यासाठी हे टप्पे फॉलो करा:


1. एक व्हर्च्युअल एन्व्हायर्नमेंट तयार करा आणि openai इन्स्टॉल करा:

  > [!NOTE] जर तुम्ही हे नोटबुक Codespaces किंवा Devcontainer मध्ये चालवत असाल तर ही पायरी आवश्यक नाही


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> तुम्ही Windows वापरत असाल तर `source venv/bin/activate` ऐवजी `venv\Scripts\activate` टाइप करा.

> [!NOTE]
> तुमची OpenAI key शोधण्यासाठी https://platform.openai.com/settings/organization/api-keys या लिंकला जा आणि `API keys` शोधा. तिथे तुम्ही नवीन key तयार करू शकता आणि लगेचच त्याची value कॉपी करू शकता.


1. *app.py* नावाची फाइल तयार करा आणि त्यात खालील कोड द्या:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## वेगवेगळ्या गोष्टींसाठी वेगवेगळ्या प्रकारचे प्रॉम्प्ट्स

आता तुम्ही पाहिलं आहे की प्रॉम्प्ट वापरून मजकूर कसा तयार करता येतो. तुमच्याकडे एक प्रोग्राम देखील आहे जो तुम्ही बदलू शकता आणि वेगवेगळ्या प्रकारचा मजकूर तयार करण्यासाठी वापरू शकता.

प्रॉम्प्ट्स वेगवेगळ्या कामांसाठी वापरता येतात. उदाहरणार्थ:

- **एखाद्या प्रकारचा मजकूर तयार करा**. उदाहरणार्थ, तुम्ही कविता, क्विझसाठी प्रश्न इ. तयार करू शकता.
- **माहिती शोधा**. तुम्ही प्रॉम्प्ट्स वापरून माहिती शोधू शकता, जसे की 'वेब डेव्हलपमेंटमध्ये CORS म्हणजे काय?'.
- **कोड तयार करा**. तुम्ही प्रॉम्प्ट्स वापरून कोड तयार करू शकता, उदाहरणार्थ ईमेल व्हॅलिडेट करण्यासाठी रेग्युलर एक्स्प्रेशन तयार करणे किंवा संपूर्ण प्रोग्राम तयार करणे, जसे की वेब अ‍ॅप?

## व्यायाम - एक रेसिपी जनरेटर तयार करा

आता आपण एक परिस्थिती पाहिली आहे, चला त्या परिस्थितीसाठी कोड लिहूया. हे करण्यासाठी, खालील पावले पाळा:

1. विद्यमान *app.py* फाइलचा प्रारंभिक बिंदू म्हणून वापरा
1. `prompt` या व्हेरिएबलला शोधा आणि त्याचा कोड खालीलप्रमाणे बदला:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

> NOTE, तुमचा LLM nondeterministic आहे, त्यामुळे प्रत्येक वेळी प्रोग्राम चालवल्यावर वेगळे परिणाम मिळू शकतात.

छान, आता पाहूया आपण हे कसे सुधारू शकतो. गोष्टी सुधारण्यासाठी, आपल्याला कोड लवचिक ठेवायचा आहे, म्हणजे घटक आणि रेसिपींची संख्या सुधारता आणि बदलता येईल.


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
