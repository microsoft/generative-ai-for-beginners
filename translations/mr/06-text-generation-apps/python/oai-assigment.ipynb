# टेक्स्ट जनरेशन अॅप्लिकेशन्स तयार करा

आपण आतापर्यंत या अभ्यासक्रमातून पाहिले आहे की प्रॉम्प्ट्स सारखे मूलभूत संकल्पना आहेत आणि अगदी "प्रॉम्प्ट इंजिनिअरिंग" नावाचा एक संपूर्ण शिस्तही आहे. अनेक साधने जसे ChatGPT, Office 365, Microsoft Power Platform आणि बरेच काही ज्यांशी आपण संवाद साधू शकता, त्यांना काहीतरी साध्य करण्यासाठी प्रॉम्प्ट्स वापरण्याचे समर्थन आहे.

आपल्याला एखाद्या अॅपमध्ये अशा प्रकारचा अनुभव समाविष्ट करायचा असल्यास, प्रॉम्प्ट्स, पूर्णता सारख्या संकल्पनांचा समज आवश्यक आहे आणि वापरण्यासाठी एक लायब्ररी निवडावी लागते. हेच आपण या प्रकरणात शिकणार आहात.

## ओळख

या प्रकरणात, आपण:

- openai लायब्ररी आणि त्याच्या मूलभूत संकल्पनांविषयी जाणून घ्या.
- openai वापरून एक टेक्स्ट जनरेशन अॅप तयार करा.
- टेक्स्ट जनरेशन अॅप तयार करण्यासाठी प्रॉम्प्ट, तापमान, टोकन्स यांसारख्या संकल्पना कशा वापरायच्या ते समजून घ्या.

## शिक्षणाचे उद्दिष्टे

या धड्याच्या शेवटी, आपण सक्षम असाल:

- टेक्स्ट जनरेशन अॅप म्हणजे काय ते स्पष्ट करा.
- openai वापरून टेक्स्ट जनरेशन अॅप तयार करा.
- आपल्या अॅपसाठी जास्त किंवा कमी टोकन्स वापरण्यासाठी आणि वेगळ्या आउटपुटसाठी तापमान बदलण्यासाठी कॉन्फिगर करा.

## टेक्स्ट जनरेशन अॅप म्हणजे काय?

सामान्यतः जेव्हा आपण अॅप तयार करता, तेव्हा त्यात खालीलप्रमाणे काही प्रकारचा इंटरफेस असतो:

- आदेश-आधारित. कन्सोल अॅप्स सामान्यतः असे असतात जिथे आपण आदेश टाइप करता आणि ते कार्य पार पाडते. उदाहरणार्थ, `git` हा एक आदेश-आधारित अॅप आहे.
- वापरकर्ता इंटरफेस (UI). काही अॅप्समध्ये ग्राफिकल वापरकर्ता इंटरफेस (GUI) असतो जिथे आपण बटणांवर क्लिक करता, मजकूर इनपुट करता, पर्याय निवडता आणि बरेच काही.

### कन्सोल आणि UI अॅप्स मर्यादित आहेत

एका आदेश-आधारित अॅपशी तुलना करा ज्यात आपण एक आदेश टाइप करता:

- **हे मर्यादित आहे**. आपण फक्त अॅप समर्थन करणारे आदेश टाइप करू शकता, कोणतेही आदेश नाही.
- **भाषा-विशिष्ट**. काही अॅप्स अनेक भाषा समर्थन करतात, पण पूर्वनिर्धारित अॅप विशिष्ट भाषेसाठी तयार केलेले असते, जरी आपण जास्त भाषा समर्थन जोडू शकता.

### टेक्स्ट जनरेशन अॅप्सचे फायदे

तर टेक्स्ट जनरेशन अॅप कसा वेगळा आहे?

टेक्स्ट जनरेशन अॅपमध्ये, आपल्याला अधिक लवचिकता असते, आपण संबद्ध आदेशांशी किंवा विशिष्ट इनपुट भाषेशी मर्यादित नाही. त्याऐवजी, आपण नैसर्गिक भाषेतून अॅपशी संवाद करू शकता. अजून एक फायदा म्हणजे आपण आधीच माहितीच्या मोठ्या डेटासेटवर प्रशिक्षण घेतलेल्या डेटा स्रोताशी संवाद साधत आहात, जिथे पारंपारिक अॅप डेटाबेसमध्ये काय आहे त्यावर मर्यादित असू शकतो.

### मी टेक्स्ट जनरेशन अॅप वापरून काय तयार करू शकतो?

आपण अनेक गोष्टी तयार करू शकता. उदाहरणार्थ:

- **चॅटबॉट**. चॅटबॉट जो आपल्या कंपनी आणि तिच्या उत्पादनांबाबत प्रश्नांची उत्तरे देतो, तो चांगला पर्याय आहे.
- **सहाय्यक**. LLMs मजकूराचा सारांश बनविणे, मजकूरातून अंतर्दृष्टी मिळविणे, रिझ्युमे सारखा मजकूर तयार करणे आणि बरेच काही करण्यात उत्कृष्ट असतात.
- **कोड सहाय्यक**. आपण वापरत असलेल्या भाषा मॉडेलवर अवलंबून, आपण कोड लिहिण्यात मदत करणारा कोड सहाय्यक तयार करू शकता. उदाहरणार्थ, GitHub Copilot सारखे उत्पादन तसेच ChatGPT वापरून कोड लिहिण्यात मदत करता येते.

## मी कसे सुरुवात करू शकतो?

बरं, आपण LLM शी एकत्रित काम कसे करायचे ते शोधावे लागेल, ज्यामध्ये सहसा खालील दोन पद्धती असतात:

- API वापरा. येथे आपण आपला प्रॉम्प्टसह वेब विनंत्या तयार करता आणि त्यातून तयार झालेला मजकूर परत प्राप्त करता.
- लायब्ररी वापरा. लायब्ररी API कॉल्सची आवरणे करतात आणि त्यांना वापरणे सोपे बनवतात.

## लायब्रऱ्या/SDKs

LLMs बाबत कार्य करण्यासाठी काही प्रसिद्ध लायब्ररी आहेत जसे की:

- **openai**, ही लायब्ररी आपल्या मॉडेलशी जोडणे आणि प्रॉम्प्ट्स पाठवणे सोपे करते.

नंतर काही उंच स्तरावर काम करणाऱ्या लायब्ररी आहेत जसे की:

- **Langchain**. Langchain प्रसिद्ध आहे आणि Python ला समर्थन देते.
- **Semantic Kernel**. Semantic Kernel हे Microsoft द्वारे तयार केलेली लायब्ररी आहे जी C#, Python, आणि Java या भाषांना समर्थन देते.

## openai वापरून पहिले अॅप

चला पाहूया की आपण पहिले अॅप कसे तयार करू शकतो, कोणत्या लायब्रऱ्या लागतात, किती आवश्यक आहे आणि इतर गोष्टी.

### openai स्थापित करा

  > [!NOTE] जर हा नोटबुक Codespaces किंवा Devcontainer मध्ये चालवत असाल तर हा टप्पा आवश्यक नाही


OpenAI किंवा Azure OpenAI सह संवाद साधण्यासाठी अनेक लायब्रऱ्या उपलब्ध आहेत. अनेक प्रोग्रामिंग भाषा जसे C#, Python, JavaScript, Java आणि बरेच काही वापरणे शक्य आहे.
आपण `openai` Python लायब्ररी वापरण्याचा निर्णय घेतला आहे, म्हणून `pip` वापरून त्याचे इंस्टॉलेशन करू.

```bash
pip install openai
```

जर आपण हा नोटबुक Codespaces किंवा Dev Container मध्ये चालवत नसाल, तर आपल्याला आपल्या मशीनवर [Python](https://www.python.org/) देखील स्थापित करावे लागेल.

### एक संसाधन तयार करा आणि आपला API की शोधा

जर आपण अद्याप हे केले नसेल, तर खालील टप्पे पूर्ण करणे आवश्यक आहे:

- OpenAI वर एक खाते तयार करा <https://platform.openai.com/signup>.
- आता, आपली API की मिळवा <https://platform.openai.com/api-keys>.

>[!NOTE]
> आपल्या API की कोडपासून वेगळे ठेवणे उपयुक्त आहे. आपण हे पर्यावरणीय बदलत्या वापरून करू शकता.
> - आपली .env फाईल मध्ये `OPENAI_KEY` या पर्यावरण बदलतात आपली API की सेट करा. जर आपण या कोर्सच्या आधीच्या सरावाने पूर्ण केले असेल, तर आपले सर्व काही तयार आहे.
> - हे लक्षात ठेवणे महत्त्वाचे आहे की API की फक्त एकदाच उपलब्ध होईल. म्हणून, ती योग्यरीतीने कॉपी केली आहे की नाही हे तपासणे आवश्यक आहे. जर ती नीट काम करत नसेल तर ती काढून टाकावी आणि नवीन की तयार करावी.


### OpenAI साठी कॉन्फिगरेशन सेट करा

आपण OpenAI वापरत असाल, तर कॉन्फिगरेशन कसे सेट करायचे:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

वर आपण खालील सेट करत आहोत:

- `api_key`, ही आपली OpenAI डॅशबोर्डमध्ये आढळणारी API की आहे.
- `deployment`, ही आपली GPT आवृत्ती आहे.

> [!NOTE]
> `os.environ` ही एक फंक्शन आहे जी पर्यावरण बदलत्या वाचते. आपण त्याचा वापर `OPENAI_API_KEY` सारखे पर्यावरण बदलत्या वाचण्यासाठी करू शकता.

## मजकूर तयार करा

मजकूर तयार करण्याचा मार्ग म्हणजे `responses.create` पद्धत वापरणे. उदाहरण येथे आहे:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

वरच्या कोडमध्ये, आपण एक प्रतिक्रिया ऑब्जेक्ट तयार करतो आणि वापरण्यासाठी हवे असलेले मॉडेल आणि प्रॉम्प्ट पास करतो. नंतर आपण तयार झालेला मजकूर छापतो.

### चॅट प्रतिसाद

आतापर्यंत, आपण पाहिले आहे की आपण एकमेव प्रॉम्प्टवरून मजकूर कसा तयार करू शकतो. Responses API चॅटबॉटसाठीही चांगले आहे - आपण संभाषण तयार करण्यासाठी `input` म्हणून संदेशांची यादी प्रदान करता. येथे एक उदाहरण आहे:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

या कार्यक्षमतेबाबत अधिक पुढील प्रकरणात.

## सराव - तुमचा पहिला टेक्स्ट जनरेशन अॅप

आता आपण OpenAI सेवा सेट अप आणि कॉन्फिगर करण्याचा मार्ग शिकलो असल्याने, आपला पहिला टेक्स्ट जनरेशन अॅप तयार करण्याची वेळ आली आहे. आपला अॅप तयार करण्यासाठी खालील टप्पे पाळा:


1. एक व्हर्च्युअल एन्व्हायर्नमेंट तयार करा आणि openai स्थापित करा:

  > [!NOTE] जर तुम्ही हा नोटबुक Codespaces किंवा Devcontainer मध्ये चालवत असाल तर हा टप्पा आवश्यक नाही


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> आपण Windows वापरत असाल तर `source venv/bin/activate` ऐवजी `venv\Scripts\activate` टाइप करा.

> [!NOTE]
> आपली OpenAI की शोधण्यासाठी https://platform.openai.com/settings/organization/api-keys येथे जा आणि `API keys` शोधा. आपण तिथे नवीन की तयार करू शकता आणि त्वरित मूल्य कॉपी करू शकता.


1. एक *app.py* फाइल तयार करा आणि त्यात खालील कोड द्या:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    तुम्हाला पुढीलप्रमाणे आउटपुट दिसेल:

    ```output
     खूप असुखी _____.

    एकेकाळी एक खूप असुखी जलपरी होती.
    ```


## विविध प्रकारच्या प्रॉम्प्ट्स, विविध कामांसाठी

आता तुम्ही पाहिले आहे की प्रॉम्प्ट वापरून मजकूर कसा तयार करायचा. तुमच्याकडे एक प्रोग्राम देखील चालू आहे ज्यामध्ये तुम्ही बदल करून वेगवेगळे प्रकारचे मजकूर जनरेट करू शकता.

प्रॉम्प्ट्स विविध कामांसाठी वापरता येतात. उदाहरणार्थ:

- **मजकूराचा एक प्रकार तयार करा**. उदाहरणार्थ, तुम्ही कविता, क्विझसाठी प्रश्न यांसारखा मजकूर तयार करू शकता.
- **माहिती शोधण्यासाठी वापरा**. तुम्ही प्रॉम्प्ट्स वापरून खालील उदाहरणासारखी माहिती शोधू शकता 'वेब विकासात CORS चा काय अर्थ आहे?'.
- **कोड तयार करा**. तुम्ही प्रॉम्प्ट्स वापरून कोड तयार करू शकता, उदाहरणार्थ ईमेलची प्रमाणीकरण करण्यासाठी रेग्युलर एक्सप्रेशन तयार करणे किंवा पूर्ण प्रोग्राम तयार करणे, जसे की वेब अॅप.

## सराव - एक रेसिपी जनरेटर तयार करा

आता जेव्हा आपण एक परिस्थिती सादर केली आहे, चला ती परिस्थिती जुळणारा कोड लिहूया. तसे करण्यासाठी, खालील चरणांचे अनुसरण करा:

1. विद्यमान *app.py* फाइलचा वापर प्रारंभ बिंदू म्हणून करा
1. `prompt` व्हेरिएबल शोधा आणि त्याचा कोड खालील प्रमाणे बदला:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

छान, चला पाहूया आपण कसे सुधारणा करू शकतो. सुधारणा करण्यासाठी, आपल्याला खात्री करायची आहे की कोड लवचीक आहे, त्यामुळे घटक आणि पाककृतींची संख्या सुधारली आणि बदली जाऊ शकते.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

आपल्याकडे आता अशी कार्यशील अॅप आहे जी पाककृती तयार करू शकते आणि ती वापरकर्त्याच्या इनपुटवर आधारित लवचिक आहे, पाककृतींच्या संख्येबरोबरच वापरलेल्या घटकांवर देखील.
