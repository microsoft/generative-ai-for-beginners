# मजकूर निर्माण अनुप्रयोग तयार करा

तुम्ही आतापर्यंत या अभ्यासक्रमामध्ये पाहिले आहे की प्रॉम्प्ट्स सारखे मूलभूत संकल्पना आहेत आणि अगदी "प्रॉम्प्ट अभियांत्रिकी" नावाचा एक संपूर्ण क्षेत्र आहे. तुम्ही ChatGPT, Office 365, Microsoft Power Platform आणि बरेच काही यांसारख्या अनेक साधनांसह संवाद साधता जे तुम्हाला काहीतरी साध्य करण्यासाठी प्रॉम्प्ट्स वापरण्याचा आधार देतात.

तुम्हाला अशाप्रकारचा अनुभव ज्याने कोणत्याही अनुप्रयोगात जोडायचा असेल तर तुम्हाला प्रॉम्प्ट्स, पूर्णता यांसारख्या संकल्पना समजून घेणे आवश्यक आहे आणि काम करण्यासाठी एखादी लायब्ररी निवडावी लागते. हेच तुम्ही या प्रकरणात शिकणार आहात.

## परिचय

या प्रकरणात, तुम्ही:

- openai लायब्ररी आणि तिच्या मूलभूत संकल्पना शिकाल.
- openai वापरून एक मजकूर निर्माण अनुप्रयोग तयार कराल.
- प्रॉम्प्ट, तापमान, आणि टोकन्स सारख्या संकल्पना कशा वापरायच्या हे समजून घेऊन मजकूर निर्माण अनुप्रयोग तयार कराल.

## शिकण्याची उद्दिष्टे

या धड्याच्या शेवटी, तुम्ही सक्षम असाल:

- मजकूर निर्माण अनुप्रयोग म्हणजे काय याचे स्पष्टीकरण द्या.
- openai वापरून मजकूर निर्माण अनुप्रयोग तयार करा.
- तुमचा अनुप्रयोग अधिक किंवा कमी टोकन्स वापरण्यास आणि विविध परिणामांसाठी तापमान बदलण्यास सेटअप करा.

## मजकूर निर्माण अनुप्रयोग म्हणजे काय?

सामान्यतः जेव्हा तुम्ही अनुप्रयोग तयार करता तेव्हा त्यात खालीलप्रमाणे काही प्रकारचा इंटरफेस असतो:

- कमांड-आधारित. कन्सोल अनुप्रयोग सामान्यतः अशा अनुप्रयोगांना म्हणतात जेथे तुम्ही एक कमांड टाइप करता आणि ते काम पूर्ण करतात. उदाहरणार्थ, `git` हा एक कमांड-आधारित अनुप्रयोग आहे.
- वापरकर्ता इंटरफेस (UI). काही अनुप्रयोगांमध्ये ग्राफिकल यूजर इंटरफेस (GUI) असतो जिथे तुम्ही बटणे क्लिक करता, मजकूर इनपुट करता, पर्याय निवडता आणि असे अनेक काही करता येते.

### कन्सोल आणि UI अनुप्रयोग मर्यादित आहेत

तुलनेत एका कमांड-आधारित अनुप्रयोगास, जिथे तुम्ही एक कमांड टाइप करता:

- **हे मर्यादित आहे**. तुम्ही कोणतीही कमांड टाइप करू शकत नाही, फक्त त्या कमांड्स ज्या अनुप्रयोगाला समर्थन आहेत तेच वापरू शकता.
- **भाषा-विशिष्ट**. काही अनुप्रयोग अनेक भाषा समर्थन करतात, पण पूर्वनिर्धारितपणे अनुप्रयोग विशिष्ट भाषेसाठी बनवलेला असतो, जरी तुम्ही अधिक भाषा समर्थन जोडू शकता.

### मजकूर निर्माण अनुप्रयोगांचे फायदे

मग मजकूर निर्माण अनुप्रयोग कसा वेगळा आहे?

मजकूर निर्माण अनुप्रयोगात तुम्हाला अधिक लवचीकता असते, तुम्ही कमांड्सच्या एका संचाबंदी किंवा विशिष्ट इनपुट भाषेपुरती सीमित नसता. त्याऐवजी तुम्ही नैसर्गिक भाषा वापरून अनुप्रयोगाशी संवाद साधू शकता. आणखी एक फायदा म्हणजे तुम्ही अशा डेटास्रोताशी आधीच संवाद साधत आहात ज्याला भरपूर माहितीवर प्रशिक्षण दिले गेले आहे, ज्यायोगे पारंपरिक अनुप्रयोगाच्या तुलनेत जे डेटाबेसमध्ये साठवलेले आहे त्यावर मर्यादित असतो.

### मजकूर निर्माण अनुप्रयोग वापरून मी काय तयार करू शकतो?

तुम्ही अनेक गोष्टी तयार करू शकता. उदाहरणार्थ:

- **चॅटबॉट**. तुमच्या कंपनी आणि तिच्या उत्पादनांविषयी प्रश्नांना उत्तर देणारा चॅटबॉट एक चांगला पर्याय असू शकतो.
- **मदतनीस**. LLMs सारखे भाषा मॉडेल्स मजकूर सारांशित करणे, मजकुरातून माहिती मिळवणे, रेस्यूमे सारख्या मजकूर तयार करणे यांसारख्या गोष्टींसाठी उपयुक्त आहेत.
- **कोड सहाय्यक**. तुम्ही वापरत असलेल्या भाषा मॉडेलवर अवलंबून, तुम्ही कोड लिहिण्यास मदत करणारा सहाय्यक तयार करू शकता. उदाहरणार्थ, तुम्ही GitHub Copilot किंवा ChatGPT सारखे उत्पादन वापरू शकता.

## मी कशी सुरुवात करू शकतो?

बरं, तुम्हाला एखाद्या LLM शी एकत्रीकरण करण्याचा मार्ग सापडावा लागेल जो साधारणतः पुढील दोन पद्धती अपयश करतो:

- API वापरा. येथे तुम्ही तुमच्या प्रॉम्प्टसह वेब विनंत्या तयार करता आणि निर्मित मजकूर प्राप्त करतो.
- लायब्ररी वापरा. लायब्ररी API कॉल्स एनकॅप्स्युलेट करण्यात मदत करतात आणि वापरण्यास सोपे करतात.

## लायब्ररी/SDKs

LLMs सोबत काम करण्यासाठी काही परिचित लायब्ररी आहेत जसे की:

- **openai**, ही लायब्ररी तुमच्या मॉडेलशी सहजपणे जोडण्यास आणि प्रॉम्प्ट्स पाठविण्यास मदत करते.

त्यानंतर काही लायब्ररी ज्या उच्च पातळीवर कार्य करतात जसे की:

- **Langchain**. Langchain चांगले परिचित आहे आणि Python ला समर्थन देते.
- **Semantic Kernel**. Semantic Kernel हा Microsoft कडून एक लायब्ररी आहे ज्याला C#, Python, आणि Java भाषांना समर्थन आहे.

## ChatCompletions वापरून मजकूर तयार करा

मजकूर तयार करण्याचा मार्ग म्हणजे `ChatCompletionsClient` वर्ग वापरणे.
`samples/python/azure_ai_inference/basic.py` मध्ये, प्रतिसादाच्या सेक्शनमध्ये, user role मध्ये खालीलप्रमाणे content parameter बदला:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # पर्यायी पॅरामिटर्
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अद्यतनित फाईल चालवा आणि आउटपुट पहा


## वेगवेगळ्या गोष्टींसाठी वेगवेगळ्या प्रकारचे प्रॉम्प्ट्स

प्रॉम्प्ट्स विविध कामांसाठी वापरले जाऊ शकतात. उदाहरणार्थ:

## व्यायाम - एक रेसिपी जनरेटर तयार करा

आता जेव्हा आपण एक परिदृश्य प्ले केले आहे, चला आपण दाखवलेल्या परिदृश्याशी जुळणारा कोड लिहूया. तसे करण्यासाठी, खालील पावले अनुसरा:

1. विद्यमान फाइल वापरून सुरुवात करा
1. `prompt` नावाचा एक व्हेरिएबल तयार करा आणि नमुना कोड खालीलप्रमाणे बदला:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

छान, चला पाहूया आपण कसे सुधार करू शकतो. सुधार करण्यासाठी, आपल्याला कोड लवचीक बनवायचा आहे, जेणेकरून घटक आणि पाककृतींची संख्या सुधारली आणि बदलली जाऊ शकतील.


1. चला खालीलप्रमाणे कोड बदला:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
