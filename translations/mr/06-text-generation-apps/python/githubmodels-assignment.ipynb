# मजकूर निर्मिती अनुप्रयोग तयार करा

तुम्ही आतापर्यंत या अभ्यासक्रमादरम्यान पाहिले आहे की प्रॉम्प्टसारखे मूलभूत संकल्पना आहेत आणि "प्रॉम्प्ट इंजिनिअरिंग" नावाचे पूर्ण क्षेत्र देखील आहे. तुम्ही ChatGPT, Office 365, Microsoft Power Platform आणि इतर अनेक साधनांसारख्या अनेक साधनांशी संवाद साधू शकता, जे काही साध्य करण्यासाठी प्रॉम्प्टचा वापर करण्यास तुम्हाला समर्थन देतात.

तुम्हाला अशा अनुभवांना अॅपमध्ये समाविष्ट करायचे असल्यास, तुम्हाला प्रॉम्प्ट, पूर्णता यांसारख्या संकल्पना समजून घ्याव्या लागतील आणि काम करण्यासाठी लायब्ररी निवडावी लागेल. याचं तुम्हाला या प्रकरणात शिकवण दिली जाईल.

## परिचय

या प्रकरणात, तुम्हाला:

- openai लायब्ररी आणि तिच्या मूलभूत संकल्पनांबद्दल शिकायचे आहे.
- openai वापरून मजकूर निर्मिती अॅप तयार करायचे आहे.
- प्रॉम्प्ट, तापमान, टोकन्स यांसारख्या संकल्पनांचा वापर करून मजकूर निर्मिती अॅप कसा तयार करायचा हे समजून घ्यायचे आहे.

## शिकण्याची उद्दिष्टे

या धड्याच्या शेवटी, तुम्ही करू शकाल:

- मजकूर निर्मिती अॅप म्हणजे काय हे स्पष्ट करणे.
- openai वापरून मजकूर निर्मिती अॅप तयार करणे.
- तुमच्या अॅपला अधिक किंवा कमी टोकन्स वापरायला कॉन्फिगर करणे आणि याशिवाय वेरिएबल आउटपुटसाठी तापमान बदलणे.

## मजकूर निर्मिती अॅप म्हणजे काय?

सामान्यतः जेव्हा तुम्ही अॅप तयार करता तेव्हा त्याचे काही प्रकारचे इंटरफेस असतात, जसे की खालीलप्रमाणे:

- कमांड-आधारित. कन्सोल अॅप्स हे असे अॅप्स आहेत ज्यांत तुम्ही कमांड टाइप करता आणि ते तेवढाच कार्य पूर्ण करतात. उदाहरणार्थ, `git` हा एक कमांड-आधारित अॅप आहे.
- युजर इंटरफेस (UI). काही अॅप्समध्ये ग्राफिकल युजर इंटरफेस (GUI) असतो, जिथे तुम्ही बटणे क्लिक करता, मजकूर टाइप करता, पर्याय निवडता आणि बरेच काही करता.

### कन्सोल आणि UI अॅप्स मर्यादित असतात

त्याची तुलना कमांड-आधारित अॅपशी करा जिथे तुम्ही कमांड टाइप करता:

- **हे मर्यादित आहे**. तुम्ही कोणतीही कमांड टाइप करू शकत नाही, फक्त ज्या कमांडसाठी अॅप समर्थन देतो त्या करता येतात.
- **भाषा-विशिष्ट**. काही अॅप्स अनेक भाषा समर्थन देतात, पण डीफॉल्टने अॅप एका विशिष्ट भाषेसाठी तयार केलेले असते, जरी तुम्ही आणखी भाषा समर्थन जोडू शकता.

### मजकूर निर्मिती अॅप्सचे फायदे

मग मजकूर निर्मिती अॅप कसे वेगळे आहे?

मजकूर निर्मिती अॅपमध्ये, तुम्हाला अधिक लवचिकता असते, तुम्ही फक्त कमांडच्या संच किंवा विशिष्ट इनपुट भाषेपुरते मर्यादित नसता. त्याऐवजी, तुम्ही नैसर्गिक भाषा वापरून अॅपशी संवाद साधू शकता. आणखी एक फायदा म्हणजे तुम्ही असे डेटा स्रोत वापरत आहात ज्यावर विशाल माहितीचा मोठा संच वापरून प्रशिक्षण झाले आहे, तर पारंपरिक अॅपला डेटाबेसमधील मर्यादित माहिती आहे.

### मजकूर निर्मिती अॅप वापरून मी काय तयार करू शकतो?

अनेक प्रकारचे काही तयार करता येतात. उदाहरणार्थ:

- **एक चॅटबॉट**. तुमच्या कंपनी आणि तिच्या उत्पादनांविषयी प्रश्नांची उत्तरे देणारा चॅटबॉट एक चांगला पर्याय असू शकतो.
- **साहाय्यक**. LLMs मजकूर सारांशित करणे, मजकूरातून अंतर्दृष्टी मिळवणे, फिरून मजकूर तयार करणे जसे की रिज्युम घेतल्यासारखे यांसारखे काम चांगल्या प्रकारे करतात.
- **कोड सहाय्यक**. तुम्ही वापरत असलेल्या भाषा मॉडेलवर अवलंबून, तुम्ही कोड लिहिण्यास मदत करणारा सहाय्यक तयार करू शकता. उदाहरणार्थ, GitHub Copilot आणि ChatGPT सारखी उत्पादने कोड लिहिण्यात मदत करू शकतात.

## सुरूवात कशी करावी?

बरं, तुम्हाला LLM सह समाकलित होण्याचा मार्ग शोधावा लागेल, जो सहसा खालील दोन पद्धती असतो:

- API वापरा. येथे तुम्ही तुमच्या प्रॉम्प्टसह वेब विनंत्या तयार करता आणि निर्मित मजकूर परत मिळवता.
- लायब्ररी वापरा. लायब्रऱ्या API कॉल्सला समाकलित करतात आणि त्याचा वापर सुलभ करतात.

## लायब्रऱ्या/SDKs

LLMs सोबत काम करण्यासाठी काही प्रसिद्ध लायब्रऱ्या आहेत जसे की:

- **openai** - ही लायब्ररी तुमच्या मॉडेलशी कनेक्ट होणे आणि प्रॉम्प्ट पाठवणे सोपे करते.

त्यानंतर काही उच्च पातळीवरील लायब्रऱ्या आहेत जसे की:

- **Langchain** - Langchain प्रसिद्ध असून Python ला समर्थन देतो.
- **Semantic Kernel** - Semantic Kernel ही Microsoft ची लायब्ररी आहे जी C#, Python, आणि Java भाषा समर्थित करते.

## ChatCompletions वापरून मजकूर तयार करा

मजकूर तयार करण्याचा मार्ग म्हणजे `ChatCompletionsClient` क्लासचा वापर करणे.
`samples/python/azure_ai_inference/basic.py` मध्ये, उत्तर भागातील कोड अपडेट करा वापरकर्त्याचा भूमिका कंटेंट पॅरामीटर खालीलप्रमाणे बदलून:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ऐच्छिक पॅरामीटर्स
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट केलेली फाईल चालवा आणि आउटपुट पहा


## वेगवेगळ्या गोष्टींसाठी वेगवेगळ्या प्रकारचे प्रॉम्प्ट्स

आता तुम्ही पाहिले आहे की प्रॉम्प्ट वापरून मजकूर कसा तयार करायचा. तुमच्याकडे एक प्रोग्राम चालू आहे ज्यात तुम्ही बदल करू शकता आणि विविध प्रकारचा मजकूर तयार करू शकता.

## व्यायाम - एक रेसिपी जनरेटर तयार करा

आता जेव्हा आपण एक घटना सादर केली आहे, तेव्हा ती सादर केलेल्या घटनेशी जुळवण्यासाठी कोड लिहूया. तसे करण्यासाठी, खालील टप्पे पाळा:

1. विद्यमान फाइल वापरा प्रारंभ बिंदू म्हणून
1. एक `prompt` व्हेरिएबल तयार करा आणि खाली नमुन्याबद्ध कोड प्रमाणे बदला:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

चांगले, चला पाहूया आपण कसे सुधारणा करू शकतो. सुधारणा करण्यासाठी, आपल्याला कोड लवचीक ठेवायचा आहे, जेणे करून साहित्य आणि पाककृतींची संख्या सुधारली आणि बदलली जाऊ शकते.


1. चला खालीलप्रमाणे कोड बदलूया:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **तापमानासोबत प्रयोग करा**. तापमान असा काहीतरी आहे ज्याचा आतापर्यंत उल्लेख केलेला नाही, पण तो आपल्या प्रोग्रामच्या कामगिरीसाठी महत्त्वाचा आहे. तापमान जास्त असेल तर आउटपुट अधिक यादृच्छिक (random) असेल. उलट, तापमान कमी असेल तर आउटपुट अधिक पूर्वानुमानित असेल. आपल्याला आपल्या आउटपुटमध्ये विविधता हवी आहे की नाही हे विचार करा.
