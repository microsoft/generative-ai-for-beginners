# टेक्स्ट जनरेशन अ‍ॅप्स तयार करा

या अभ्यासक्रमात तुम्ही आतापर्यंत पाहिलं आहे की प्रॉम्प्टसारख्या मूलभूत संकल्पना आहेत आणि "प्रॉम्प्ट इंजिनिअरिंग" नावाचं एक संपूर्ण क्षेत्र आहे. ChatGPT, Office 365, Microsoft Power Platform आणि इतर अनेक टूल्समध्ये तुम्ही प्रॉम्प्ट वापरून काहीतरी साध्य करू शकता.

तुमच्या अ‍ॅपमध्ये अशी सुविधा जोडायची असेल, तर तुम्हाला प्रॉम्प्ट, कम्प्लिशन्स यांसारख्या संकल्पना समजून घ्याव्या लागतील आणि कोणती लायब्ररी वापरायची ते ठरवावं लागेल. ह्याच गोष्टी तुम्ही या अध्यायात शिकणार आहात.

## परिचय

या अध्यायात तुम्ही:

- openai लायब्ररी आणि तिच्या मुख्य संकल्पना शिकाल.
- openai वापरून टेक्स्ट जनरेशन अ‍ॅप तयार कराल.
- प्रॉम्प्ट, टेम्परेचर आणि टोकन्ससारख्या संकल्पना वापरून टेक्स्ट जनरेशन अ‍ॅप कसं तयार करायचं हे समजून घ्याल.

## शिकण्याची उद्दिष्टे

या धड्याच्या शेवटी, तुम्ही हे करू शकाल:

- टेक्स्ट जनरेशन अ‍ॅप म्हणजे काय हे समजावून सांगू शकाल.
- openai वापरून टेक्स्ट जनरेशन अ‍ॅप तयार करू शकाल.
- तुमच्या अ‍ॅपमध्ये जास्त किंवा कमी टोकन्स वापरण्यासाठी आणि वेगळ्या आउटपुटसाठी टेम्परेचर बदलण्यासाठी सेटिंग्ज करू शकाल.

## टेक्स्ट जनरेशन अ‍ॅप म्हणजे काय?

साधारणपणे जेव्हा तुम्ही अ‍ॅप तयार करता, तेव्हा त्याला खालीलपैकी काही इंटरफेस असतो:

- कमांड-आधारित. कन्सोल अ‍ॅप्स म्हणजे जिथे तुम्ही कमांड टाकता आणि ते काम करतात. उदाहरणार्थ, `git` हे कमांड-आधारित अ‍ॅप आहे.
- युजर इंटरफेस (UI). काही अ‍ॅप्सना ग्राफिकल युजर इंटरफेस (GUI) असतो जिथे तुम्ही बटणं क्लिक करता, मजकूर टाकता, पर्याय निवडता इ.

### कन्सोल आणि UI अ‍ॅप्स मर्यादित असतात

कमांड-आधारित अ‍ॅपशी तुलना करा:

- **मर्यादित आहे**. तुम्ही कोणतीही कमांड टाकू शकत नाही, फक्त अ‍ॅप जे सपोर्ट करतं तेवढ्याच.
- **भाषा-विशिष्ट**. काही अ‍ॅप्स अनेक भाषा सपोर्ट करतात, पण मूळतः ते एका विशिष्ट भाषेसाठी तयार केलेले असतात, जरी तुम्ही इतर भाषा जोडू शकत असलात तरी.

### टेक्स्ट जनरेशन अ‍ॅप्सचे फायदे

मग टेक्स्ट जनरेशन अ‍ॅप वेगळं कसं आहे?

टेक्स्ट जनरेशन अ‍ॅपमध्ये तुम्हाला जास्त लवचिकता मिळते, तुम्ही फक्त ठराविक कमांड किंवा इनपुट भाषेपुरते मर्यादित राहत नाही. त्याऐवजी, तुम्ही नैसर्गिक भाषेतून अ‍ॅपशी संवाद साधू शकता. आणखी एक फायदा म्हणजे, तुम्ही अशा डेटासोर्सशी संवाद साधत आहात ज्यावर प्रचंड माहितीवर ट्रेनिंग झालं आहे, तर पारंपारिक अ‍ॅपमध्ये डेटाबेसमध्ये जेवढं आहे तेवढ्यापुरतंच मर्यादित असू शकतं.

### टेक्स्ट जनरेशन अ‍ॅपने काय तयार करू शकतो?

तुम्ही बरीच वेगवेगळी अ‍ॅप्स तयार करू शकता. उदाहरणार्थ:

- **चॅटबॉट**. एखाद्या विषयावर प्रश्नांची उत्तरं देणारा चॅटबॉट, जसं की तुमच्या कंपनीबद्दल किंवा प्रॉडक्ट्सबद्दल, हे चांगलं उदाहरण आहे.
- **सहाय्यक**. LLMs मजकूर संक्षिप्त करणे, मजकुरातून माहिती मिळवणे, रिझ्युमेसारखा मजकूर तयार करणे अशा गोष्टींमध्ये खूप चांगले आहेत.
- **कोड असिस्टंट**. तुम्ही कोणता भाषा मॉडेल वापरता यावर अवलंबून, तुम्ही कोड लिहायला मदत करणारा कोड असिस्टंट तयार करू शकता. उदाहरणार्थ, GitHub Copilot किंवा ChatGPT वापरून तुम्ही कोड लिहायला मदत मिळवू शकता.

## सुरुवात कशी करावी?

तुम्हाला LLM शी कनेक्ट होण्याचा मार्ग शोधावा लागेल, जो साधारणपणे दोन प्रकारे करता येतो:

- API वापरा. इथे तुम्ही प्रॉम्प्टसह वेब रिक्वेस्ट तयार करता आणि त्यावरून जनरेट केलेला मजकूर मिळवता.
- लायब्ररी वापरा. लायब्ररी API कॉल्स सहज वापरता येतील अशा पद्धतीने तयार केलेल्या असतात.

## लायब्ररी/SDKs

LLMsसाठी काही प्रसिद्ध लायब्ररी आहेत:

- **openai**, ही लायब्ररी तुमच्या मॉडेलशी कनेक्ट होऊन प्रॉम्प्ट पाठवणं सोपं करते.

याशिवाय, काही उच्च-स्तरीय लायब्ररी आहेत:

- **Langchain**. Langchain प्रसिद्ध आहे आणि Python ला सपोर्ट करते.
- **Semantic Kernel**. Semantic Kernel ही Microsoft ची लायब्ररी आहे, जी C#, Python आणि Java या भाषांना सपोर्ट करते.

## ChatCompletions वापरून टेक्स्ट जनरेट करा

टेक्स्ट जनरेट करण्यासाठी `ChatCompletionsClient` क्लास वापरावा लागतो.
`samples/python/azure_ai_inference/basic.py` मध्ये, response सेक्शनमध्ये, user role साठी content parameter खालीलप्रमाणे बदला:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

अपडेट केलेली फाईल चालवा आणि आउटपुट पाहा


## वेगवेगळ्या गोष्टींसाठी वेगवेगळ्या प्रकारचे प्रॉम्प्ट्स

आता तुम्ही पाहिलं आहे की प्रॉम्प्ट वापरून मजकूर कसा तयार करता येतो. तुमच्याकडे एक प्रोग्राम देखील आहे जो तुम्ही बदलू शकता आणि वेगवेगळ्या प्रकारचा मजकूर तयार करण्यासाठी वापरू शकता.

प्रॉम्प्ट्स वेगवेगळ्या कामांसाठी वापरता येतात. उदाहरणार्थ:

- **एखाद्या प्रकारचा मजकूर तयार करा**. उदाहरणार्थ, तुम्ही कविता, क्विझसाठी प्रश्न वगैरे तयार करू शकता.
- **माहिती शोधा**. तुम्ही प्रॉम्प्ट्स वापरून माहिती शोधू शकता, जसं की 'वेब डेव्हलपमेंटमध्ये CORS म्हणजे काय?'.
- **कोड तयार करा**. तुम्ही प्रॉम्प्ट्स वापरून कोड तयार करू शकता, जसं की ईमेल व्हॅलिडेट करण्यासाठी रेग्युलर एक्स्प्रेशन किंवा संपूर्ण प्रोग्राम, जसं की वेब अ‍ॅप तयार करणे.

## सराव: एक रेसिपी जनरेटर

कल्पना करा की तुमच्याकडे घरी काही साहित्य आहे आणि तुम्हाला काहीतरी बनवायचं आहे. त्यासाठी तुम्हाला रेसिपी लागेल. रेसिपी शोधण्याचा एक मार्ग म्हणजे सर्च इंजिन वापरणे किंवा LLM वापरणे.

## व्यायाम - एक रेसिपी जनरेटर तयार करा

आता आपण एक परिस्थिती पाहिली आहे, चला त्या परिस्थितीसाठी कोड लिहूया. हे करण्यासाठी, खालील पावले पाळा:

1. विद्यमान फाइलचा प्रारंभिक बिंदू म्हणून वापर करा
1. एक `prompt` नावाचा व्हेरिएबल तयार करा आणि नमुना कोड खालीलप्रमाणे बदला:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

छान, आता पाहूया आपण हे कसे सुधारू शकतो. गोष्टी सुधारण्यासाठी, आपल्याला कोड लवचिक ठेवायचा आहे, म्हणजे घटक आणि रेसिपींची संख्या सहज बदलता येईल आणि सुधारता येईल.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- "स्टडी बडी" तयार करा. हे अ‍ॅप एखाद्या विषयावर प्रश्नांची उत्तरे देऊ शकेल, उदाहरणार्थ Python, तुम्ही असे प्रॉम्प्ट्स देऊ शकता "Python मधील एखादा विषय काय आहे?", किंवा "मला त्या विषयासाठी कोड दाखवा" वगैरे.
- हिस्ट्री बॉट, इतिहास जिवंत करा, बॉटला एखाद्या ऐतिहासिक पात्राची भूमिका द्या आणि त्याच्या आयुष्याबद्दल प्रश्न विचारा.
