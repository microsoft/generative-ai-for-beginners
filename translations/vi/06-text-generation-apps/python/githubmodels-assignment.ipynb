# Xây dựng ứng dụng tạo văn bản

Bạn đã thấy trong chương trình học này có những khái niệm cốt lõi như prompts và thậm chí một lĩnh vực hoàn toàn gọi là "kỹ thuật prompt". Nhiều công cụ bạn có thể tương tác như ChatGPT, Office 365, Microsoft Power Platform và hơn thế nữa, hỗ trợ bạn sử dụng prompts để hoàn thành một việc gì đó.

Để bạn thêm trải nghiệm như vậy vào một ứng dụng, bạn cần hiểu các khái niệm như prompts, completions và chọn một thư viện để làm việc. Đó chính xác là những gì bạn sẽ học trong chương này.

## Giới thiệu

Trong chương này, bạn sẽ:

- Tìm hiểu về thư viện openai và các khái niệm cốt lõi của nó.
- Xây dựng một ứng dụng tạo văn bản sử dụng openai.
- Hiểu cách sử dụng các khái niệm như prompt, nhiệt độ và tokens để xây dựng ứng dụng tạo văn bản.

## Mục tiêu học tập

Cuối bài học này, bạn sẽ có thể:

- Giải thích ứng dụng tạo văn bản là gì.
- Xây dựng ứng dụng tạo văn bản sử dụng openai.
- Cấu hình ứng dụng của bạn để sử dụng nhiều hoặc ít tokens hơn và thay đổi nhiệt độ, cho ra kết quả đa dạng.

## Ứng dụng tạo văn bản là gì?

Thông thường khi bạn xây dựng một ứng dụng, nó có một giao diện dạng như sau:

- Dựa trên lệnh. Ứng dụng console là các ứng dụng điển hình nơi bạn nhập lệnh và ứng dụng thực hiện nhiệm vụ đó. Ví dụ, `git` là ứng dụng dựa trên lệnh.
- Giao diện người dùng (UI). Một số ứng dụng có giao diện người dùng đồ họa (GUI) nơi bạn nhấp nút, nhập văn bản, chọn tùy chọn và hơn thế nữa.

### Ứng dụng console và UI bị giới hạn

So sánh với ứng dụng dựa trên lệnh nơi bạn nhập một lệnh:

- **Nó bị giới hạn**. Bạn không thể nhập bất kỳ lệnh nào, chỉ các lệnh mà ứng dụng hỗ trợ.
- **Ngôn ngữ cụ thể**. Một số ứng dụng hỗ trợ nhiều ngôn ngữ, nhưng theo mặc định ứng dụng được xây dựng cho một ngôn ngữ cụ thể, dù bạn có thể thêm hỗ trợ ngôn ngữ khác.

### Lợi ích của ứng dụng tạo văn bản

Vậy ứng dụng tạo văn bản khác như thế nào?

Trong một ứng dụng tạo văn bản, bạn có nhiều linh hoạt hơn, bạn không bị giới hạn bởi một tập lệnh lệnh hoặc một ngôn ngữ đầu vào cụ thể. Thay vào đó, bạn có thể sử dụng ngôn ngữ tự nhiên để tương tác với ứng dụng. Một lợi ích khác là vì bạn đang tương tác với nguồn dữ liệu đã được huấn luyện trên một kho tàng thông tin khổng lồ, trong khi một ứng dụng truyền thống có thể bị giới hạn bởi dữ liệu trong cơ sở dữ liệu.

### Tôi có thể xây dựng gì với ứng dụng tạo văn bản?

Có nhiều thứ bạn có thể xây dựng. Ví dụ:

- **Chatbot**. Chatbot trả lời câu hỏi về các chủ đề, như công ty của bạn và sản phẩm của nó có thể là một lựa chọn phù hợp.
- **Trợ lý**. Các LLM rất giỏi trong các việc như tổng hợp văn bản, lấy thông tin từ văn bản, tạo văn bản như hồ sơ cá nhân và hơn thế nữa.
- **Trợ lý code**. Tùy thuộc vào mô hình ngôn ngữ bạn sử dụng, bạn có thể xây dựng trợ lý code giúp bạn viết mã. Ví dụ, bạn có thể sử dụng sản phẩm như GitHub Copilot cũng như ChatGPT để giúp viết mã.

## Làm thế nào để bắt đầu?

- Sử dụng thư viện. Các thư viện giúp đóng gói các lời gọi API và làm cho chúng dễ sử dụng hơn.

## Bài tập - xây dựng trình tạo công thức

Bây giờ chúng ta đã chơi một kịch bản, hãy viết mã để khớp với kịch bản đã trình bày. Để làm vậy, hãy thực hiện các bước sau:

1. Sử dụng tệp hiện có làm điểm bắt đầu
1. Tạo biến `prompt` và thay đổi mã mẫu như bên dưới:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Tuyệt vời, hãy cùng xem cách chúng ta có thể cải thiện. Để cải thiện, chúng ta muốn đảm bảo mã có tính linh hoạt, vì vậy các nguyên liệu và số lượng công thức có thể được cải thiện và thay đổi.


1. Hãy thay đổi mã theo cách sau đây:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
