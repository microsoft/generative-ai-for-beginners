# Xây dựng ứng dụng tạo văn bản  

Bạn đã thấy qua chương trình học này rằng có những khái niệm cốt lõi như prompt và thậm chí cả một lĩnh vực gọi là "kỹ thuật prompt". Nhiều công cụ bạn có thể tương tác như ChatGPT, Office 365, Microsoft Power Platform và nhiều hơn nữa, hỗ trợ bạn sử dụng prompt để hoàn thành một việc gì đó.  

Để bạn có thể thêm trải nghiệm như vậy vào một ứng dụng, bạn cần hiểu các khái niệm như prompt, completions và chọn một thư viện để làm việc. Đó chính là những gì bạn sẽ học trong chương này.  

## Giới thiệu  

Trong chương này, bạn sẽ:  

- Tìm hiểu về thư viện openai và các khái niệm cốt lõi của nó.  
- Xây dựng một ứng dụng tạo văn bản sử dụng openai.  
- Hiểu cách sử dụng các khái niệm như prompt, temperature và tokens để xây dựng ứng dụng tạo văn bản.  

## Mục tiêu học tập  

Cuối bài học này, bạn sẽ có thể:  

- Giải thích được ứng dụng tạo văn bản là gì.  
- Xây dựng được ứng dụng tạo văn bản sử dụng openai.  
- Cấu hình ứng dụng sử dụng nhiều hay ít tokens đồng thời thay đổi nhiệt độ (temperature) để có đầu ra đa dạng.  

## Ứng dụng tạo văn bản là gì?  

Thông thường khi bạn xây dựng một ứng dụng sẽ có một dạng giao diện nào đó như sau:  

- Dựa trên lệnh (Command-based). Ứng dụng console là loại ứng dụng bạn gõ lệnh và nó thực hiện công việc. Ví dụ, `git` là ứng dụng dạng command-based.  
- Giao diện người dùng (User interface - UI). Một số ứng dụng có giao diện người dùng đồ họa (GUI) nơi bạn click nút, nhập văn bản, chọn lựa chọn,...  

### Ứng dụng console và UI có nhiều giới hạn  

So sánh với ứng dụng dựa trên lệnh, bạn gõ một lệnh:  

- **Có giới hạn**. Bạn không thể tự do gõ bất kỳ lệnh nào, chỉ lệnh mà ứng dụng hỗ trợ mới thực thi được.  
- **Ngôn ngữ cụ thể**. Một số ứng dụng hỗ trợ nhiều ngôn ngữ, nhưng về mặc định ứng dụng được xây dựng cho một ngôn ngữ cụ thể, mặc dù bạn có thể thêm hỗ trợ ngôn ngữ khác.  

### Lợi ích của ứng dụng tạo văn bản  

Vậy ứng dụng tạo văn bản khác như thế nào?  

Ở ứng dụng tạo văn bản, bạn có nhiều linh hoạt hơn, không bị giới hạn trong một tập hợp lệnh hay ngôn ngữ đầu vào cụ thể. Thay vào đó, bạn có thể dùng ngôn ngữ tự nhiên để tương tác với ứng dụng. Một lợi ích khác là bạn đang tương tác với một nguồn dữ liệu đã được huấn luyện trên một kho tàng thông tin lớn, trong khi ứng dụng truyền thống có thể bị giới hạn bởi dữ liệu trong cơ sở dữ liệu.  

### Tôi có thể xây dựng gì với ứng dụng tạo văn bản?  

Có rất nhiều thứ bạn có thể xây dựng. Ví dụ:  

- **Chatbot**. Một chatbot trả lời câu hỏi về các chủ đề, chẳng hạn như công ty và sản phẩm của bạn có thể là một lựa chọn phù hợp.  
- **Trợ lý**. Các mô hình ngôn ngữ lớn (LLM) rất giỏi trong việc tóm tắt văn bản, lấy thông tin từ văn bản, tạo văn bản như hồ sơ cá nhân và nhiều hơn nữa.  
- **Trợ lý lập trình**. Tùy vào mô hình ngôn ngữ bạn dùng, bạn có thể xây dựng trợ lý lập trình giúp bạn viết code. Ví dụ, bạn có thể dùng sản phẩm như GitHub Copilot cũng như ChatGPT để giúp bạn viết code.  

## Tôi có thể bắt đầu như thế nào?

### Cấu hình OpenAI  

Nếu bạn dùng OpenAI, đây là cách cấu hình:  

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```
  
Phía trên chúng ta đang thiết lập:  

- `api_key`, đây là khóa API của bạn tìm thấy trong bảng điều khiển OpenAI.  
- `deployment`, đây là phiên bản GPT của bạn.  

> [!NOTE]  
> `os.environ` là một hàm đọc biến môi trường. Bạn có thể dùng nó để đọc biến môi trường như `OPENAI_API_KEY`.  

## Tạo văn bản  

Cách tạo văn bản là dùng phương thức `responses.create`. Đây là ví dụ:  

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```
  
Trong đoạn mã trên, chúng ta tạo một đối tượng response và truyền vào mô hình muốn sử dụng cùng với prompt. Sau đó in ra văn bản được tạo.  

### Phản hồi chat  

Cho đến nay, bạn đã thấy cách tạo văn bản từ một prompt duy nhất. API Responses cũng rất phù hợp cho chatbot - bạn cung cấp một danh sách các tin nhắn làm `input` để tạo nên cuộc hội thoại. Đây là ví dụ sử dụng:  

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```
  
Sẽ có thêm thông tin về tính năng này trong các chương tiếp theo.

1. Tạo một môi trường ảo và cài đặt openai:

## Bài tập - xây dựng trình tạo công thức

Bây giờ chúng ta đã chơi xong một kịch bản, hãy viết mã để khớp với kịch bản đã trình diễn. Để làm điều đó, hãy làm theo các bước sau:

1. Sử dụng tệp *app.py* hiện có làm điểm khởi đầu
1. Tìm biến `prompt` và thay đổi mã của nó thành như sau:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Tuyệt vời, hãy xem cách chúng ta có thể cải thiện mọi thứ. Để cải thiện, chúng ta muốn đảm bảo mã linh hoạt, để nguyên liệu và số lượng công thức có thể được nâng cấp và thay đổi.


1. Hãy thay đổi mã theo cách sau:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Việc chạy thử đoạn mã có thể trông như sau:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

### Cải thiện bằng cách thêm bộ lọc và danh sách mua sắm

Chúng ta hiện có một ứng dụng hoạt động có khả năng tạo công thức nấu ăn và nó linh hoạt vì dựa trên các đầu vào từ người dùng, cả về số lượng công thức cũng như nguyên liệu được sử dụng.

Để cải thiện thêm, chúng ta muốn thêm những điều sau:

- **Lọc nguyên liệu**. Chúng ta muốn có thể lọc bỏ những nguyên liệu mà chúng ta không thích hoặc bị dị ứng. Để thực hiện thay đổi này, chúng ta có thể chỉnh sửa prompt hiện có và thêm điều kiện lọc vào cuối prompt như sau:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Ở trên, chúng ta thêm `{filter}` vào cuối prompt và cũng nhận giá trị lọc từ người dùng.

    Một ví dụ đầu vào khi chạy chương trình bây giờ có thể trông như sau:

    Như bạn thấy, bất kỳ công thức nào có sữa đều bị lọc ra. Nhưng nếu bạn bị dị ứng lactose, bạn có thể muốn lọc cả công thức có phô mai, vì vậy cần phải rõ ràng.

- **Tạo danh sách mua sắm**. Chúng ta muốn tạo danh sách mua sắm, xem xét những gì đã có sẵn trong nhà.

    Cho chức năng này, chúng ta có thể thử giải quyết mọi thứ trong một prompt hoặc chúng ta có thể chia nó thành hai prompt. Hãy thử cách thứ hai. Ở đây chúng ta đề xuất thêm một prompt bổ sung, nhưng để điều đó hoạt động, chúng ta cần thêm kết quả của prompt đầu tiên làm ngữ cảnh cho prompt thứ hai.

    Tìm phần trong mã in ra kết quả từ prompt đầu tiên và thêm mã sau phía dưới:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # in phản hồi
    print("Shopping list:")
    print(response.output_text)
    ```

    Lưu ý các điểm sau:

    - Chúng ta xây dựng một prompt mới bằng cách thêm kết quả từ prompt đầu tiên vào prompt mới:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Chúng ta thực hiện một yêu cầu mới, nhưng cũng xét đến số lượng tokens đã yêu cầu trong prompt đầu tiên, vì vậy lần này chúng ta nói `max_output_tokens` là 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```

- **Lời nói về độ dài token**. Chúng ta nên cân nhắc số lượng token cần dùng để tạo ra văn bản mong muốn. Tokens sẽ mất phí, vì vậy nếu có thể, chúng ta nên tiết kiệm số lượng token sử dụng. Ví dụ, liệu chúng ta có thể diễn đạt prompt sao cho dùng ít token hơn không?

   Để thay đổi số token sử dụng, bạn có thể dùng tham số `max_output_tokens`. Ví dụ, nếu bạn muốn dùng 100 token, bạn sẽ làm như sau:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Thử nghiệm với nhiệt độ (temperature)**. Nhiệt độ là một yếu tố mà chúng ta chưa đề cập đến cho đến bây giờ nhưng là bối cảnh quan trọng cho hoạt động của chương trình. Giá trị nhiệt độ càng cao thì đầu ra càng ngẫu nhiên. Ngược lại, nhiệt độ càng thấp thì đầu ra càng dự đoán được. Hãy xem bạn có muốn sự đa dạng trong đầu ra hay không.

   Để thay đổi nhiệt độ, bạn có thể dùng tham số `temperature`. Ví dụ, nếu muốn dùng nhiệt độ 0.5, bạn sẽ làm:

