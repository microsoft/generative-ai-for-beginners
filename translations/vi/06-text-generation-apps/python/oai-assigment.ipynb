# Xây dựng ứng dụng tạo văn bản

Bạn đã thấy cho đến nay trong chương trình này có những khái niệm cốt lõi như prompt và thậm chí một lĩnh vực riêng gọi là "kỹ thuật prompt". Nhiều công cụ bạn có thể tương tác như ChatGPT, Office 365, Microsoft Power Platform và nhiều hơn nữa, hỗ trợ bạn sử dụng prompt để hoàn thành một việc gì đó.

Để bạn có thể thêm trải nghiệm như vậy vào một ứng dụng, bạn cần hiểu các khái niệm như prompt, completions và chọn một thư viện để làm việc. Đó chính xác là những gì bạn sẽ học trong chương này.

- Hiểu cách sử dụng các khái niệm như prompt, nhiệt độ (temperature), và tokens để xây dựng ứng dụng tạo văn bản.

## Mục tiêu học tập

## Ứng dụng tạo văn bản là gì?

Thông thường khi bạn xây dựng ứng dụng, nó có một loại giao diện như sau:

- Dựa trên lệnh. Ứng dụng console là những ứng dụng điển hình nơi bạn nhập lệnh và nó thực hiện nhiệm vụ. Ví dụ, `git` là ứng dụng dựa trên lệnh.
- Giao diện người dùng (UI). Một số ứng dụng có giao diện đồ họa người dùng (GUI) nơi bạn nhấn nút, nhập văn bản, chọn tùy chọn và nhiều hơn nữa.

### Ứng dụng console và UI có giới hạn

So sánh với ứng dụng dựa trên lệnh, nơi bạn nhập một lệnh:

- **Nó bị giới hạn**. Bạn không thể nhập bất cứ lệnh nào, chỉ các lệnh mà ứng dụng hỗ trợ.
- **Truyền ngôn ngữ cụ thể**. Một số ứng dụng hỗ trợ nhiều ngôn ngữ, nhưng mặc định ứng dụng thường được xây dựng cho một ngôn ngữ cụ thể, mặc dù có thể thêm hỗ trợ nhiều ngôn ngữ.

### Lợi ích của ứng dụng tạo văn bản

Vậy ứng dụng tạo văn bản khác biệt như thế nào?

Trong ứng dụng tạo văn bản, bạn có nhiều sự linh hoạt hơn, bạn không bị giới hạn vào tập lệnh hoặc ngôn ngữ đầu vào cụ thể. Thay vào đó, bạn có thể dùng ngôn ngữ tự nhiên để tương tác với ứng dụng. Một lợi ích khác là vì bạn đang tương tác với nguồn dữ liệu đã được huấn luyện trên một tập hợp lớn thông tin, trong khi ứng dụng truyền thống có thể bị giới hạn vào dữ liệu có trong cơ sở dữ liệu.

### Tôi có thể xây dựng gì với ứng dụng tạo văn bản?

Có rất nhiều thứ bạn có thể xây dựng. Ví dụ:

- **Chatbot**. Một chatbot trả lời các câu hỏi về các chủ đề, ví dụ như công ty và sản phẩm của bạn có thể là sự lựa chọn tốt.
- **Trợ giúp**. Các mô hình ngôn ngữ lớn (LLMs) rất giỏi những việc như tóm tắt văn bản, rút ra thông tin từ văn bản, tạo các văn bản như sơ yếu lý lịch và hơn thế nữa.
- **Trợ lý mã**. Tùy thuộc vào mô hình ngôn ngữ bạn sử dụng, bạn có thể xây dựng trợ lý mã giúp bạn viết code. Ví dụ, bạn có thể sử dụng sản phẩm như GitHub Copilot cũng như ChatGPT để giúp viết code.

## Tôi có thể bắt đầu như thế nào?

Bạn cần tìm cách tích hợp với một mô hình ngôn ngữ lớn (LLM) thường bao gồm hai phương pháp sau:

- Sử dụng API. Ở đây, bạn xây dựng các yêu cầu web với prompt của bạn và nhận lại văn bản được tạo ra.
- Sử dụng thư viện. Các thư viện giúp đóng gói các cuộc gọi API và làm cho chúng dễ sử dụng hơn.

### Cấu hình OpenAI

Nếu bạn dùng OpenAI, đây là cách thiết lập cấu hình:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

Ở trên, chúng ta thiết lập các giá trị sau:

- `api_key`, đây là khóa API của bạn tìm thấy trong bảng điều khiển OpenAI.
- `deployment`, đây là phiên bản GPT của bạn.

> [!NOTE]
> `os.environ` là một hàm đọc biến môi trường. Bạn có thể dùng nó để đọc biến môi trường như `OPENAI_API_KEY`.

## Tạo văn bản

Cách tạo văn bản là dùng phương thức `responses.create`. Dưới đây là ví dụ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Trong đoạn mã trên, chúng ta tạo một đối tượng response và truyền mô hình muốn dùng và prompt. Sau đó in ra văn bản được tạo.

### Phản hồi trò chuyện

Cho đến nay, bạn đã thấy cách tạo văn bản từ một prompt đơn lẻ. API Responses cũng rất phù hợp cho chatbot — bạn cung cấp danh sách các tin nhắn làm `input` để xây dựng cuộc hội thoại. Dưới đây là ví dụ sử dụng:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Thêm về chức năng này trong chương sắp tới.

1. Tạo một môi trường ảo và cài đặt openai:

## Bài tập - xây dựng bộ tạo công thức nấu ăn

Bây giờ chúng ta đã mô phỏng một kịch bản, hãy viết mã để khớp với kịch bản đã trình bày. Để làm được điều đó, hãy làm theo các bước sau:

1. Sử dụng tệp *app.py* hiện có làm điểm bắt đầu
1. Tìm biến `prompt` và thay đổi mã của nó thành như sau:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Tuyệt vời, hãy xem cách chúng ta có thể cải thiện mọi thứ. Để cải thiện mọi thứ, chúng ta muốn đảm bảo mã linh hoạt, vì vậy các thành phần và số lượng công thức có thể được cải thiện và thay đổi.


1. Hãy thay đổi mã theo cách sau:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Chạy thử mã, có thể trông như sau:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

### Cải thiện bằng cách thêm bộ lọc và danh sách mua sắm

Chúng ta hiện có một ứng dụng hoạt động có khả năng tạo ra các công thức và nó linh hoạt vì dựa trên dữ liệu đầu vào từ người dùng, cả về số lượng công thức lẫn nguyên liệu sử dụng.

Để cải thiện hơn nữa, chúng ta muốn thêm các điều sau:

- **Lọc nguyên liệu**. Chúng ta muốn có thể lọc bỏ những nguyên liệu không thích hoặc dị ứng. Để thực hiện thay đổi này, chúng ta có thể chỉnh sửa prompt hiện tại và thêm điều kiện lọc vào cuối như sau:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Phía trên, chúng ta thêm `{filter}` vào cuối prompt và đồng thời thu thập giá trị bộ lọc từ người dùng.

    Ví dụ khi chạy chương trình bây giờ có thể trông như sau:
    

    Như bạn thấy, bất kỳ công thức nào có sữa đã bị lọc ra. Nhưng, nếu bạn bị không dung nạp lactose, bạn cũng có thể muốn lọc bỏ các công thức có phô mai, vì vậy cần phải rõ ràng.

- **Tạo danh sách mua sắm**. Chúng ta muốn tạo danh sách mua sắm, dựa trên những gì đã có ở nhà.

    Đối với chức năng này, chúng ta có thể thử giải quyết tất cả trong một prompt hoặc chia thành hai prompt. Hãy thử cách thứ hai. Ở đây chúng tôi đề xuất thêm một prompt bổ sung, nhưng để cách này hoạt động, chúng ta cần đưa kết quả của prompt trước làm ngữ cảnh cho prompt sau.

    Xác định phần trong mã in ra kết quả từ prompt đầu tiên và thêm mã sau đây bên dưới:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # in phản hồi
    print("Shopping list:")
    print(response.output_text)
    ```

    Lưu ý các điểm sau:

    - Chúng ta xây dựng một prompt mới bằng cách thêm kết quả từ prompt đầu tiên vào prompt mới:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Chúng ta thực hiện yêu cầu mới, đồng thời cân nhắc số lượng token đã yêu cầu trong prompt đầu tiên, lần này chúng ta đặt `max_output_tokens` là 1200.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Chạy đoạn mã này, ta thu được kết quả sau:

- **Chú ý về độ dài token**. Chúng ta nên xem xét cần bao nhiêu token để tạo ra văn bản mong muốn. Token tốn tiền, vì vậy nếu có thể, nên cố gắng tiết kiệm số token dùng. Ví dụ, ta có thể chỉnh lại prompt sao cho dùng bớt token không?

   Để thay đổi token sử dụng, bạn có thể dùng tham số `max_output_tokens`. Ví dụ, nếu bạn muốn dùng 100 token, bạn sẽ làm như sau:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Thử nghiệm với nhiệt độ (temperature)**. Nhiệt độ là điều chúng ta chưa nhắc đến nhưng rất quan trọng đối với cách chương trình vận hành. Nhiệt độ càng cao, kết quả càng ngẫu nhiên. Nhiệt độ càng thấp, kết quả càng dự đoán được. Cân nhắc bạn có muốn đầu ra đa dạng hay không.

   Để thay đổi nhiệt độ, bạn có thể dùng tham số `temperature`. Ví dụ, nếu bạn muốn dùng nhiệt độ 0.5, bạn làm như sau:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Lưu ý, càng gần 1.0 thì kết quả càng đa dạng.
