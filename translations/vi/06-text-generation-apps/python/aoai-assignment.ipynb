# Xây dựng ứng dụng tạo văn bản

Bạn đã thấy cho đến nay trong chương trình này có những khái niệm cốt lõi như prompt và thậm chí một lĩnh vực chuyên ngành gọi là "kỹ thuật prompt". Nhiều công cụ bạn có thể tương tác như ChatGPT, Office 365, Microsoft Power Platform và nhiều hơn nữa, hỗ trợ bạn sử dụng prompt để hoàn thành một việc gì đó.

Để bạn có thể thêm trải nghiệm đó vào một ứng dụng, bạn cần hiểu các khái niệm như prompt, hoàn thành và chọn một thư viện để làm việc. Đó chính xác là những gì bạn sẽ học trong chương này.

## Giới thiệu

- Hiểu cách sử dụng các khái niệm như prompt, temperature và tokens để xây dựng ứng dụng tạo văn bản.

## Mục tiêu học tập

Vào cuối bài học này, bạn sẽ có thể:

## Ứng dụng tạo văn bản là gì?

Thông thường khi bạn xây dựng một ứng dụng, nó có một loại giao diện như sau:

- Dựa trên lệnh. Ứng dụng console là ứng dụng điển hình nơi bạn nhập một lệnh và nó thực hiện một nhiệm vụ. Ví dụ, `git` là một ứng dụng dựa trên lệnh.
- Giao diện người dùng (UI). Một số ứng dụng có giao diện đồ họa người dùng (GUI) nơi bạn nhấn nút, nhập văn bản, chọn tùy chọn và nhiều hơn nữa.

### Ứng dụng Console và UI bị giới hạn

So sánh với một ứng dụng dựa trên lệnh nơi bạn nhập lệnh:

- **Nó bị giới hạn**. Bạn không thể nhập bất kỳ lệnh nào, chỉ những lệnh mà ứng dụng hỗ trợ.
- **Ngôn ngữ cụ thể**. Một số ứng dụng hỗ trợ nhiều ngôn ngữ, nhưng theo mặc định ứng dụng được xây dựng cho một ngôn ngữ cụ thể, ngay cả khi bạn có thể thêm hỗ trợ ngôn ngữ khác.

### Lợi ích của ứng dụng tạo văn bản

Vậy ứng dụng tạo văn bản khác như thế nào?

Trong một ứng dụng tạo văn bản, bạn có nhiều linh hoạt hơn, bạn không bị giới hạn trong một tập lệnh hay ngôn ngữ đầu vào cụ thể. Thay vào đó, bạn có thể sử dụng ngôn ngữ tự nhiên để tương tác với ứng dụng. Một lợi ích khác là vì bạn đang tương tác với một nguồn dữ liệu đã được huấn luyện trên một khối lượng lớn thông tin, trong khi một ứng dụng truyền thống có thể bị giới hạn trong dữ liệu có trong cơ sở dữ liệu.

### Tôi có thể xây dựng gì với ứng dụng tạo văn bản?

Có nhiều thứ bạn có thể xây dựng. Ví dụ:

- **Chatbot**. Một chatbot trả lời câu hỏi về các chủ đề, như công ty bạn và các sản phẩm của nó có thể là lựa chọn phù hợp.
- **Trợ lý**. Các LLM rất giỏi trong các việc như tóm tắt văn bản, lấy thông tin từ văn bản, tạo văn bản như CV và nhiều hơn nữa.
- **Trợ lý lập trình**. Tùy vào mô hình ngôn ngữ bạn sử dụng, bạn có thể xây dựng trợ lý giúp bạn viết mã. Ví dụ, bạn có thể sử dụng sản phẩm như GitHub Copilot cũng như ChatGPT để giúp bạn viết mã.

### Cấu hình Azure

Nếu bạn sử dụng Azure OpenAI, đây là cách thiết lập cấu hình. Responses API được phục vụ từ endpoint Azure OpenAI **v1**, vì vậy ta chỉ định client `OpenAI` tại `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Ở trên chúng ta đang thiết lập:

- `api_key`, đây là khoá API bạn tìm thấy trong Azure Portal.
- `base_url`, đây là endpoint Azure OpenAI của bạn với `/openai/v1/` được thêm vào. Bạn có thể tìm endpoint trong Azure Portal bên cạnh khoá API. Sử dụng endpoint v1 có nghĩa là bạn không cần phải truyền `api_version`.
- `deployment`, đây là tên triển khai mô hình bạn đã tạo trong portal Foundry.

> [!NOTE]
> `os.environ` là một ánh xạ đọc các biến môi trường. Bạn có thể dùng nó để đọc biến môi trường như `AZURE_OPENAI_API_KEY` và `AZURE_OPENAI_ENDPOINT`.

## Tạo văn bản

Cách để tạo văn bản là sử dụng phương thức `responses.create`. Đây là ví dụ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Trong đoạn mã trên, chúng ta tạo một đối tượng response và truyền vào mô hình muốn sử dụng và prompt. Sau đó chúng ta in ra văn bản đã tạo.

### Phản hồi Chat

Responses API rất phù hợp cho cả tạo văn bản theo lượt đơn lẻ và chatbot theo nhiều lượt - bạn chỉ cần cung cấp thêm tin nhắn vào danh sách `input` để xây dựng cuộc hội thoại:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Sẽ có thêm phần về chức năng này trong chương tới.

## Bài tập - ứng dụng tạo văn bản đầu tiên của bạn

Bây giờ bạn đã học cách thiết lập và cấu hình dịch vụ Azure OpenAI, đã đến lúc xây dựng ứng dụng tạo văn bản đầu tiên của bạn. Để xây dựng ứng dụng, hãy làm theo các bước sau:


1. Tạo một môi trường ảo và cài đặt openai:

  > [!NOTE] Bước này không cần thiết nếu bạn chạy sổ tay này trên Codespaces hoặc trong một Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> Nếu bạn đang sử dụng Windows, hãy nhập `venv\Scripts\activate` thay vì `source venv/bin/activate`. 

> [!NOTE]
> Tìm khóa Azure OpenAI của bạn bằng cách truy cập https://portal.azure.com/ và tìm kiếm `Open AI` sau đó chọn `Open AI resource` và tiếp theo chọn `Keys and Endpoint` rồi sao chép giá trị `Key 1`.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## Bài tập - xây dựng trình tạo công thức

Bây giờ chúng ta đã chơi thử một kịch bản, hãy viết mã để khớp với kịch bản đã trình diễn. Để làm như vậy, hãy làm theo các bước sau:

1. Sử dụng tệp *app.py* hiện có làm điểm bắt đầu
1. Tìm biến `prompt` và thay đổi mã của nó thành như sau:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200.

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Note, the closer to 1.0, the more varied the output.
