# ਇਮેજ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ (OpenAI)

LLMs ਵੱਖ-ਵੱਖ ਤਰ੍ਹਾਂ ਦੇ ਹਨ, ਸਿਰਫ਼ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਹੀ ਨਹੀਂ। ਤੁਸੀਂ ਟੈਕਸਟ ਵਰਣਨ ਤੋਂ ਚਿੱਤਰ ਵੀ ਜਨਰੇਟ ਕਰ ਸਕਦੇ ਹੋ। ਚਿੱਤਰਾਂ ਇੱਕ ਮੋਡੈਲਿਟੀ ਵਜੋਂ MedTech, ਵਾਸਤੁਕਲਾ, ਸੈਰ-ਸਪਾਟਾ, ਗੇਮ ਵਿਕਾਸ, ਮਾਰਕੀਟਿੰਗ, ਅਤੇ ਹੋਰ ਖੇਤਰਾਂ ਵਿੱਚ ਮਦਦਗਾਰ ਹਨ। ਇਸ ਪਾਠ ਵਿੱਚ ਅਸੀਂ ਅੱਜ ਦੇ **GPT ਇਮೇಜ** ਮਾਡਲਾਂ ਨਾਲ OpenAI ਪਲੇਟਫਾਰਮ 'ਤੇ ਕੰਮ ਕਰਾਂਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕਸ਼ਿਆ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਤੱਕ ਤੁਸੀਂ ਸਮਰੱਥ ਹੋ ਜाओਗੇ:

- ਸਮਝਾ ਸਕੋ ਕਿ ਇਮેજ ਜਨਰੇਸ਼ਨ ਕੀ ਹੈ ਅਤੇ ਇਹ ਕਿੱਥੇ ਲਾਭਦਾਇਕ ਹੈ।
- `gpt-image` ਮਾਡਲ ਪਰਿਵਾਰ ਨੂੰ ਸਮਝੋ ਅਤੇ ਇਹ ਪੁਰਾਣੇ DALL·E ਮਾਡਲਾਂ ਤੋਂ ਕਿਸ ਤਰ੍ਹਾਂ ਵੱਖਰਾ ਹੈ।
- ਇੱਕ ਇਮেজ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਓ ਅਤੇ ਚਿੱਤਰਾਂ ਨੂੰ ਸੋਧੋ।

## ਇਮেজ ਜਨਰੇਸ਼ਨ ਕੀ ਹੈ?

ਇਮেজ ਜਨਰੇਸ਼ਨ ਮਾਡਲ ਟੈਕਸਟ ਪ੍ਰਾਂਪਟ ਤੋਂ ਚਿੱਤਰ ਬਣਾਉਂਦੇ ਹਨ। ਅਧੁਨਿਕ ਮਾਡਲ ਜਿਵੇਂ ਕਿ `gpt-image` ਪ੍ਰਸ਼ਿਸ਼ਣ ਦੌਰਾਨ ਟੈਕਸਟ ਅਤੇ ਚਿੱਤਰਾਂ ਦੇ ਰਿਸ਼ਤੇ ਨੂੰ ਸਿੱਖਦੇ ਹਨ, ਫਿਰ ਕਦਮ ਦਰ ਕਦਮ ਬੇਰੁਪ ਧੁਨੀ ਨੂੰ ਤੁਹਾਡੇ ਪ੍ਰਾਂਪਟ ਨਾਲ ਮੇਲ ਖਾਂਦਾ ਚਿੱਤਰ ਬਣਾਉਂਦੇ ਹਨ।

ਦੋ ਪ੍ਰਸਿੱਧ ਇਮેજ ਮਾਡਲ ਪਰਿਵਾਰ ਹਨ:

- **`gpt-image` (OpenAI)** — ਇਸ ਪਾਠ ਵਿੱਚ ਵਰਤੀ ਜਾਂਦੀ ਮਾਜ਼ੂਦਾ ਪੀੜ੍ਹੀ। ਇਹ ਟੈਕਸਟ-ਤੋਂ-ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਅਤੇ ਚਿੱਤਰ ਸੋਧ (ਮਾਸਕ ਨਾਲ ਇਨਪੇਂਟਿੰਗ) ਨੂੰ ਸਮਰਥਨ ਕਰਦਾ ਹੈ।
- **Midjourney** — ਆਪਣੀ ਸੇਵਾ ਅਤੇ Discord-ਅਧਾਰਤ ਵਰਕਫਲੋਅ ਵਾਲਾ ਪ੍ਰਸਿੱਧ ਤੀਜਾ-ਪੱਖੀ ਮਾਡਲ।

> ਪੁਰਾਣੇ OpenAI ਇਮેજ ਮਾਡਲ — **DALL·E 2** ਅਤੇ **DALL·E 3** — ਪੁਰਾਣੇ ਹਨ। DALL·E 3 ਹਣਕਾਰੀਤ ਤੌਰ 'ਤੇ ਨਵੀਆਂ ਡਿਪਲੋਇਮੈਂਟਾਂ ਲਈ ਉਪਲਬਧ ਨਹੀਂ ਹੈ, ਅਤੇ `create_variation` ਵਰਗੀਆਂ ਵਿਸ਼ੇਸ਼ਤਾਵਾਂ ਸਿਰਫ਼ DALL·E 2 ਵਿੱਚ ਮਿਲਦੀਆਂ ਸਨ। ਨਵੀਆਂ ਐਪਲੀਕੇਸ਼ਨਾਂ ਲਈ `gpt-image` ਮਾਡਲਾਂ ਨੂੰ ਵਰਤੋਂ।

> **ਮਹੱਤਵਪੂਰਣ:** `gpt-image` ਮਾਡਲ ਜਨਰੇਟ ਕੀਤਾ ਚਿੱਤਰ **base64** (`b64_json`) ਦੇ ਤੌਰ 'ਤੇ ਵਾਪਸ ਕਰਦੇ ਹਨ, URL ਵਜੋਂ ਨਹੀਂ। ਤੁਹਾਡਾ ਕੋਡ base64 ਸਤਰ ਨੂੰ ਬਾਈਟਸ ਵਿੱਚ ਡੀਕੋਡ ਕਰਦਾ ਹੈ ਅਤੇ ਉਸ ਨੂੰ ਸੇਵ ਕਰਦਾ ਹੈ — ਕੋਈ ਚਿੱਤਰ URL ਡਾਊਨਲੋਡ ਕਰਨ ਲਈ ਨਹੀਂ ਹੁੰਦਾ।


## ਆਪਣੀ ਪਹਿਲੀ ਚਿੱਤਰ ਬਣਾਉਣ ਵਾਲੀ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ

ਤਾਂ ਚਿੱਤਰ ਬਣਾਉਣ ਵਾਲੀ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣ ਲਈ ਕੀ ਚਾਹੀਦਾ ਹੈ? ਤੁਹਾਨੂੰ ਹੇਠ ਲਿਖੀਆਂ ਲਾਇਬਰੇਰੀਜ਼ ਦੀ ਲੋੜ ਹੈ:

- **python-dotenv**, ਤੁਹਾਨੂੰ ਇਹ ਲਾਇਬਰੇਰੀ ਆਪਣੇ ਸਿੱਕਰੈਟਸ ਨੂੰ ਕੋਡ ਤੋਂ ਦੂਰ *.env* ਫਾਈਲ ਵਿੱਚ ਰੱਖਣ ਲਈ ਬਹੁਤ ਸਿਫ਼ਾਰਸ਼ ਕੀਤੀ ਜਾਂਦੀ ਹੈ।
- **openai**, ਇਹ ਲਾਇਬਰੇਰੀ ਉਹ ਹੈ ਜਿਸ ਨਾਲ ਤੁਸੀਂ OpenAI API ਨਾਲ ਸੰਪਰਕ ਕਰੋਗੇ।
- **pillow**, ਫਾਈਥਨ ਵਿੱਚ ਚਿੱਤਰਾਂ ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ।


1. ਇੱਕ ਫਾਈਲ *.env* ਬਣਾਓ ਜਿਸ ਵਿੱਚ ਹੇਠ ਲਿਖਿਆ ਸਮੱਗਰੀ ਹੋਵੇ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. ਉਪਰ ਦਿੱਤੀਆਂ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਨੂੰ ਇੱਕ ਫਾਇਲ *requirements.txt* ਵਿੱਚ ਇਸ ਤਰ੍ਹਾਂ ਇਕੱਤਰ ਕਰੋ:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. ਫਿਰ, ਵਰਚੁਅਲ ਵਾਤਾਵਰਣ ਬਣਾਓ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਇੰਸਟਾਲ ਕਰੋ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Windows ਲਈ, ਆਪਣਾ ਵਰਚੁਅਲ ماحول ਬਣਾਉਣ ਅਤੇ ਐਕਟੀਵੇਟ ਕਰਨ ਲਈ ਹੇਠਾਂ ਦਿੱਤੇ ਕਮਾਂਡਾਂ ਦੀ ਵਰਤੋਂ ਕਰੋ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ਨਾਮਕ ਫਾਇਲ ਵਿੱਚ ਹੇਠਾਂ ਦਿੱਤਾ ਕੋਡ ਸ਼ਾਮਿਲ ਕਰੋ:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv.import ਕਰੋ
    dotenv.load_dotenv()

    # OpenAI ਆਬਜੈਕਟ ਬਣਾਓ (.env ਤੋਂ OPENAI_API_KEY ਪੜ੍ਹਦਾ ਹੈ)
    client = openai.OpenAI()


    try:
        # ਇਮੇਜ ਜੈਨਰੇਸ਼ਨ API ਵਰਤ ਕੇ ਇੱਕ ਚਿੱਤਰ ਬਣਾਓ
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣਾ ਪ੍ਰੋੰਪਟ ਟੈਕਸਟ ਇੱਥੇ ਦਾਖਲ ਕਰੋ
            size='1024x1024',
            n=1
        )
        # ਸਟੋਰ ਕੀਤੇ ਚਿੱਤਰ ਲਈ ਡਾਇਰੈਕਟਰੀ ਸੈੱਟ ਕਰੋ
        image_dir = os.path.join(os.curdir, 'images')

        # ਜੇ ਡਾਇਰੈਕਟਰੀ ਮੌਜੂਦ ਨਹੀਂ ਹੈ ਤਾਂ ਇਸਨੂੰ ਬਣਾਓ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ਚਿੱਤਰ ਦਾ ਪਾਥ ਇਨੀਸ਼ੀਅਲਾਈਜ਼ ਕਰੋ (ਧਿਆਨ ਰਹੇ ਫਾਇਲਟਾਈਪ png ਹੋਣਾ ਚਾਹੀਦਾ ਹੈ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image ਮਾਡਲ ਚਿੱਤਰ ਨੂੰ base64 (b64_json) ਵਜੋਂ ਵਾਪਸ ਦਿੰਦੇ ਹਨ, URL ਨਹੀਂ
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ਡਿਫਾਲਟ ਇਮੇਜ ਵਿਊਅਰ ਵਿੱਚ ਚਿੱਤਰ ਦਿਖਾਓ
        image = Image.open(image_path)
        image.show()

    # exceptions ਨੂੰ ਫੜੋ
    except openai.BadRequestError as err:
        print(err)

    ```

ਆਓ ਇਸ ਕੋਡ ਦੀ ਵਿਆਖਿਆ ਕਰੀਏ:

- ਸਭ ਤੋਂ ਪਹਿਲਾਂ, ਅਸੀਂ ਉਹ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਪੋਰਟ ਕਰਦੇ ਹਾਂ ਜਿਹੜੀਆਂ ਸਾਨੂੰ ਚਾਹੀਦੀਆਂ ਹਨ, ਜਿਵੇਂ ਕਿ OpenAI ਲਾਇਬ੍ਰੇਰੀ, dotenv ਲਾਇਬ੍ਰੇਰੀ, base64 ਮੋਡਿਊਲ, ਅਤੇ Pillow ਲਾਇਬ੍ਰੇਰੀ.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- ਉਸ ਤੋਂ ਬਾਅਦ, ਅਸੀਂ ਕਲਾਇਂਟ ਬਣਾਉਂਦੇ ਹਾਂ, ਜੋ ਤੁਹਾਡੇ ``.env`` ਤੋਂ API ਕੁੰਜੀ ਪੜ੍ਹਦਾ ਹੈ.

    ```python
    # OpenAI ਆਬਜੈਕਟ ਬਣਾਓ
    client = openai.OpenAI()
    ```

- ਫਿਰ, ਅਸੀਂ ਚਿੱਤਰ ਤਿਆਰ ਕਰਦੇ ਹਾਂ:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣੇ ਪ੍ਰੰਪਟ ਦਾ ਟੈਕਸਟ ਇੱਥੇ ਦਰਜ ਕਰੋ
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` ਮਾਡਲ ਚਿੱਤਰ ਨੂੰ **base64** ਸਤਰ ਦੇ ਤੌਰ ਤੇ `data[0].b64_json` ਵਿੱਚ ਵਾਪਸ ਕਰਦੇ ਹਨ। ਅਸੀਂ ਇਸ ਨੂੰ ਬਾਈਟਸ ਵਿੱਚ ਡੀਕੋਡ ਕਰਦੇ ਹਾਂ ਅਤੇ ਫਾਇਲ ਵਿੱਚ ਲਿਖਦੇ ਹਾਂ — ਡਾਊਨਲੋਡ ਕਰਨ ਲਈ ਕੋਈ URL ਨਹੀਂ ਹੁੰਦਾ।

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- ਆਖਿਰ ਵਿੱਚ, ਅਸੀਂ ਚਿੱਤਰ ਖੋਲ੍ਹਦੇ ਹਾਂ ਅਤੇ ਇਸਨੂੰ ਦਿਖਾਉਣ ਲਈ ਸਟੈਂਡਰਡ ਇਮੇਜ ਵੇਵਿਊਅਰ ਦੀ ਵਰਤੋਂ ਕਰਦੇ ਹਾਂ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਬਾਰੇ ਹੋਰ ਜਾਣਕਾਰੀ

ਅਸੀਂ `images.generate` ਦੇ ਪੈਰਾਮੀਟਰਾਂ ‘ਤੇ ਨਜ਼ਰ ਮਾਰਦੇ ਹਾਂ:

- **model**, ਵਰਤਣ ਵਾਲਾ ਇਮੇਜ ਮਾਡਲ (ਜਿਵੇਂ ਕਿ `gpt-image-1`).
- **prompt**, ਚਿੱਤਰ ਬਣਾਉਣ ਲਈ ਵਰਤੀ ਜਾਣ ਵਾਲੀ ਟੈਕਸਟ ਪ੍ਰਾਂਪਟ। ਇੱਥੇ ਇਹ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ਹੈ।
- **size**, ਬਣਾਏ ਗਏ ਚਿੱਤਰ ਦਾ ਆਕਾਰ (ਜਿਵੇਂ ਕਿ `1024x1024`, `1536x1024`, `1024x1536`, ਜਾਂ `"auto"`).
- **n**, ਬਣਾਏ ਗਏ ਚਿੱਤਰਾਂ ਦੀ ਗਿਣਤੀ। ਇੱਥੇ ਅਸੀਂ ਇੱਕ ਚਿੱਤਰ ਬਣਾਉਂਦੇ ਹਾਂ।

> ਇਮੇਜ ਮਾਡਲ ‘temperature’ ਪੈਰਾਮੀਟਰ ਨਹੀਂ ਲੈਂਦੇ — ਇਹ ਟੈਕਸਟ-ਜਨਰੇਸ਼ਨ ਕੰਟਰੋਲ ਹੁੰਦਾ ਹੈ। ਵੱਖ-ਵੱਖ ਪ੍ਰਕਾਰ ਪਾਉਣ ਲਈ API ਨੂੰ ਫਿਰ ਤੋਂ ਕਾਲ ਕਰੋ; ਵੱਖ-ਵੱਖਤਾ ਘਟਾਉਣ ਲਈ ਆਪਣਾ ਪ੍ਰਾਂਪਟ ਹੋਰ ਵਿਸ਼ੇਸ਼ ਬਣਾਓ।

## ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਦੀਆਂ ਹੋਰ ਖੂਬੀਆਂ

ਤੁਸੀਂ ਵੇਖਿਆ ਕਿ ਕੁਝ ਪਾਇਥਨ ਲਾਈਨਾਂ ਨਾਲ ਕਿਵੇਂ ਚਿੱਤਰ ਬਣਾਇਆ ਜਾ ਸਕਦਾ ਹੈ। `gpt-image` ਮਾਡਲ ਮੌਜੂਦਾ ਚਿੱਤਰ ਨੂੰ **ਸੰਪਾਦਿਤ** ਵੀ ਕਰ ਸਕਦੇ ਹਨ। ਇੱਕ ਚਿੱਤਰ, ਇਕ ਵਿਕਲਪਿਕ **ਮਾਸਕ** (ਜੋ ਬਦਲਾਅ ਵਾਲਾ ਖੇਤਰ ਦਿਖਾਉਂਦਾ ਹੈ), ਅਤੇ ਪ੍ਰਾਂਪਟ ਦੇ ਕੇ, ਤੁਸੀਂ ਚਿੱਤਰ ਦਾ ਹਿੱਸਾ ਬਦਲ ਸਕਦੇ ਹੋ — ਉਦਾਹਰਨ ਵਜੋਂ, ਸਾਡੇ ਖਰਗੋਸ਼ ਨੂੰ ਟੋਪੀ ਪਛਾਣ ਲਈ.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ਸੋਧਾਂ ਵੀ base64 ਵਜੋਂ ਵਾਪਸ ਕੀਤੀਆਂ ਜਾਂਦੀਆਂ ਹਨ
edited_image = base64.b64decode(response.data[0].b64_json)
```

ਬੇਸਿਕ ਚਿੱਤਰ ਵਿੱਚ ਸਿਰਫ ਖਰਗੋਸ਼ ਹੈ; ਆਖਰੀ ਚਿੱਤਰ ਵਿੱਚ ਖਰਗੋਸ਼ ਤੇ ਟੋਪੀ ਹੈ।
