## ਆਪਣਾ ਪਹਿਲਾ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ

ਤਾਂ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣ ਲਈ ਤੁਹਾਨੂੰ ਕੀ ਚਾਹੀਦਾ ਹੈ? ਤੁਹਾਨੂੰ ਇਹ ਲਾਇਬ੍ਰੇਰੀਆਂ ਦੀ ਲੋੜ ਹੋਵੇਗੀ:

- **python-dotenv**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਵਰਤਣ ਦੀ ਸਿਫਾਰਸ਼ ਕੀਤੀ ਜਾਂਦੀ ਹੈ ਤਾਂ ਜੋ ਆਪਣੇ ਰਾਜ਼ *.env* ਫਾਇਲ ਵਿੱਚ ਕੋਡ ਤੋਂ ਵੱਖ ਰੱਖ ਸਕੋ।
- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਨੂੰ OpenAI API ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਵਰਤਣੀ ਪਵੇਗੀ।
- **pillow**, Python ਵਿੱਚ ਚਿੱਤਰਾਂ ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ।
- **requests**, HTTP ਰਿਕਵੈਸਟ ਭੇਜਣ ਵਿੱਚ ਮਦਦ ਲਈ।

1. ਇੱਕ *.env* ਫਾਇਲ ਬਣਾਓ, ਜਿਸ ਵਿੱਚ ਇਹ ਸਮੱਗਰੀ ਹੋਵੇ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```


1. ਉਪਰੋਕਤ ਲਾਇਬ੍ਰੇਰੀਆਂ ਨੂੰ *requirements.txt* ਨਾਂ ਦੀ ਫਾਈਲ ਵਿੱਚ ਇਸ ਤਰ੍ਹਾਂ ਇਕੱਠਾ ਕਰੋ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```

1. ਹੁਣ, ਵਰਚੁਅਲ ਇਨਵਾਇਰਨਮੈਂਟ ਬਣਾਓ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਸਟਾਲ ਕਰੋ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

1. *app.py* ਨਾਂ ਦੀ ਫਾਈਲ ਵਿੱਚ ਹੇਠਾਂ ਦਿੱਤਾ ਕੋਡ ਪਾਓ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # import dotenv
    dotenv.load_dotenv()

    # Create OpenAI object
    client = OpenAI()


    try:
        # Create an image by using the image generation API
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
            size='1024x1024',
            n=1
        )
        # Set the directory for the stored image
        image_dir = os.path.join(os.curdir, 'images')

        # If the directory doesn't exist, create it
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Initialize the image path (note the filetype should be png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # Retrieve the generated image
        print(generation_response)

        image_url = generation_response.data[0].url # extract image URL from response
        generated_image = requests.get(image_url).content  # download the image
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Display the image in the default image viewer
        image = Image.open(image_path)
        image.show()

    # catch exceptions
    except client.error.InvalidRequestError as err:
        print(err)

    ```

    ਉਪਰੋਕਤ ਕੋਡ ਇੱਕ JSON ਆਬਜੈਕਟ ਦੇ ਰੂਪ ਵਿੱਚ ਜਵਾਬ ਦਿੰਦਾ ਹੈ, ਜਿਸ ਵਿੱਚ ਬਣੀ ਹੋਈ ਇਮੇਜ ਦਾ URL ਹੁੰਦਾ ਹੈ। ਅਸੀਂ ਇਸ URL ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇਮੇਜ ਡਾਊਨਲੋਡ ਕਰ ਸਕਦੇ ਹਾਂ ਅਤੇ ਫਾਈਲ ਵਿੱਚ ਸੇਵ ਕਰ ਸਕਦੇ ਹਾਂ।

- ਆਖਰ ਵਿੱਚ, ਅਸੀਂ ਇਮੇਜ ਖੋਲ੍ਹਦੇ ਹਾਂ ਅਤੇ ਸਟੈਂਡਰਡ ਇਮੇਜ ਵਿਊਅਰ ਰਾਹੀਂ ਵਿਖਾਉਂਦੇ ਹਾਂ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### ਇਮੇਜ ਬਣਾਉਣ ਬਾਰੇ ਹੋਰ ਜਾਣਕਾਰੀ

ਆਓ ਉਸ ਕੋਡ ਨੂੰ ਹੋਰ ਵਿਸਥਾਰ ਵਿੱਚ ਵੇਖੀਏ ਜੋ ਇਮੇਜ ਬਣਾਉਂਦਾ ਹੈ:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Enter your prompt text here
        size='1024x1024',
        n=1
    )
```

- **prompt**, ਉਹ ਟੈਕਸਟ ਪ੍ਰੋੰਪਟ ਹੈ ਜੋ ਇਮੇਜ ਬਣਾਉਣ ਲਈ ਵਰਤਿਆ ਜਾਂਦਾ ਹੈ। ਇਸ ਮਾਮਲੇ ਵਿੱਚ, ਅਸੀਂ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ਵਰਤ ਰਹੇ ਹਾਂ।
- **size**, ਇਹ ਬਣੀ ਹੋਈ ਇਮੇਜ ਦਾ ਆਕਾਰ ਹੈ। ਇੱਥੇ, ਅਸੀਂ 1024x1024 ਪਿਕਸਲ ਦੀ ਇਮੇਜ ਬਣਾਉਂਦੇ ਹਾਂ।
- **n**, ਇਹ ਬਣੀਆਂ ਹੋਈਆਂ ਇਮੇਜਾਂ ਦੀ ਗਿਣਤੀ ਹੈ। ਇੱਥੇ, ਅਸੀਂ ਦੋ ਇਮੇਜਾਂ ਬਣਾਉਂਦੇ ਹਾਂ।

ਇਮੇਜਾਂ ਨਾਲ ਹੋਰ ਵੀ ਕਈ ਕੰਮ ਕੀਤੇ ਜਾ ਸਕਦੇ ਹਨ, ਜੋ ਅਸੀਂ ਅਗਲੇ ਭਾਗ ਵਿੱਚ ਕਵਰ ਕਰਾਂਗੇ।

## ਇਮੇਜ ਜਨਰੇਸ਼ਨ ਦੀਆਂ ਵਾਧੂ ਸਮਰੱਥਾਵਾਂ

ਹੁਣ ਤੱਕ ਤੁਸੀਂ ਵੇਖਿਆ ਕਿ ਅਸੀਂ ਕੁਝ ਲਾਈਨਾਂ ਦੇ ਕੋਡ ਨਾਲ Python ਵਿੱਚ ਇਮੇਜ ਬਣਾਈ। ਪਰ, ਇਮੇਜਾਂ ਨਾਲ ਹੋਰ ਵੀ ਕਈ ਕੰਮ ਕੀਤੇ ਜਾ ਸਕਦੇ ਹਨ।

ਤੁਸੀਂ ਇਹ ਵੀ ਕਰ ਸਕਦੇ ਹੋ:

- **ਸੰਪਾਦਨ ਕਰੋ**। ਜੇ ਤੁਸੀਂ ਮੌਜੂਦਾ ਇਮੇਜ, ਇੱਕ ਮਾਸਕ ਅਤੇ ਇੱਕ ਪ੍ਰੋੰਪਟ ਦਿੰਦੇ ਹੋ, ਤਾਂ ਤੁਸੀਂ ਇਮੇਜ ਵਿੱਚ ਤਬਦੀਲੀ ਕਰ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਲਈ, ਤੁਸੀਂ ਇਮੇਜ ਦੇ ਕਿਸੇ ਹਿੱਸੇ ਵਿੱਚ ਕੁਝ ਸ਼ਾਮਲ ਕਰ ਸਕਦੇ ਹੋ। ਜਿਵੇਂ ਕਿ ਸਾਡੀ bunny ਵਾਲੀ ਇਮੇਜ, ਤੁਸੀਂ bunny ਨੂੰ ਟੋਪੀ ਪਾ ਸਕਦੇ ਹੋ। ਇਹ ਕਰਨ ਲਈ, ਤੁਸੀਂ ਇਮੇਜ, ਮਾਸਕ (ਜੋ ਹਿੱਸਾ ਬਦਲਣਾ ਹੈ) ਅਤੇ ਟੈਕਸਟ ਪ੍ਰੋੰਪਟ ਦਿੰਦੇ ਹੋ ਕਿ ਕੀ ਕਰਨਾ ਹੈ।

    ```python
    response = openai.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ਮੂਲ ਇਮੇਜ ਵਿੱਚ ਸਿਰਫ਼ ਖਰਗੋਸ਼ ਹੋਵੇਗਾ, ਪਰ ਆਖਰੀ ਇਮੇਜ ਵਿੱਚ ਖਰਗੋਸ਼ ਦੇ ਸਿਰ 'ਤੇ ਟੋਪੀ ਹੋਵੇਗੀ।
    
- **ਵੈਰੀਏਸ਼ਨ ਬਣਾਓ**। ਇਹਦਾ ਮਤਲਬ ਹੈ ਕਿ ਤੁਸੀਂ ਮੌਜੂਦਾ ਇਮੇਜ ਲੈਂਦੇ ਹੋ ਅਤੇ ਉਸ ਦੀਆਂ ਵੱਖ-ਵੱਖ ਕਿਸਮਾਂ ਬਣਾਉਂਦੇ ਹੋ। ਵੈਰੀਏਸ਼ਨ ਬਣਾਉਣ ਲਈ, ਤੁਸੀਂ ਇਮੇਜ ਅਤੇ ਟੈਕਸਟ ਪ੍ਰੋੰਪਟ ਦਿੰਦੇ ਹੋ ਅਤੇ ਹੇਠਾਂ ਦਿੱਤੇ ਤਰੀਕੇ ਨਾਲ ਕੋਡ ਲਿਖਦੇ ਹੋ:

    ```python
    response = openai.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
