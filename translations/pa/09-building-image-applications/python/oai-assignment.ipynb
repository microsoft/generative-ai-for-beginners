## ਆਪਣੀ ਪਹਿਲੀ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ

ਤਾਂ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣ ਲਈ ਕੀ ਚਾਹੀਦਾ ਹੈ? ਤੁਹਾਨੂੰ ਹੇਠਾਂ ਦਿੱਤੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਦੀ ਲੋੜ ਹੈ:

- **python-dotenv**, ਤੁਹਾਨੂੰ ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਇਸ ਲਈ ਵਰਤਣ ਦੀ ਕੜੀ ਸਿਫਾਰਸ਼ ਕੀਤੀ ਜਾਂਦੀ ਹੈ ਤਾਂ ਜੋ ਤੁਹਾਡੇ ਰਹਸਵੀਆਂ *.env* ਫਾਈਲ ਵਿੱਚ ਕੋਡ ਤੋਂ ਦੂਰ ਰਹਿਣ।
- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ ਤੁਸੀਂ OpenAI API ਨਾਲ ਗੱਲਬਾਤ ਕਰਨ ਲਈ ਵਰਤੋਂਗੇ।
- **pillow**, ਪਾਇਥਨ ਵਿੱਚ ਚਿੱਤਰਾਂ ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ।
- **requests**, HTTP ਅਨੁਰੋਧ ਬਣਾਉਣ ਵਿੱਚ ਤੁਹਾਡੇ ਦੀ ਸਹਾਇਤਾ ਲਈ।


1. ਹੇਠਾਂ ਦਿੱਤੇ ਸਮੱਗਰੀ ਨਾਲ *.env* ਨਾਂ ਦੀ ਫਾਈਲ ਬਣਾਓ:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. ਉੱਪਰ ਦਿੱਤੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਨੂੰ *requirements.txt* ਨਾਮਕ ਫਾਈਲ ਵਿੱਚ ਇਕੱਠਾ ਕਰੋ ਇਸ ਤਰ੍ਹਾਂ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. ਅਗਲਾ, ਵਰਚੁਅਲ ਵਾਤਾਵਰਨ ਬਣਾਓ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਸਟਾਲ ਕਰੋ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows ਲਈ, ਆਪਣਾ ਵਰਚੁਅਲ ਵਾਤਾਵਰਣ ਬਣਾਉਣ ਅਤੇ ਐਕਟੀਵੇਟ ਕਰਨ ਲਈ ਹੇਠ ਲਿਖੇ ਕਮਾਂਡਾਂ ਵਰਤੋਂ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* ਨਾਂ ਦੇ ਫਾਈਲ ਵਿੱਚ ਹੇਠ ਲਿਖਿਆ ਕੋਡ ਸ਼ਾਮਲ ਕਰੋ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ਇੰਪੋਰਟ ਕਰੋ
    dotenv.load_dotenv()

    # OpenAI ਆਬਜੈਕਟ ਬਣਾਓ
    client = openai.OpenAI()


    try:
        # ਇਮੇਜ ਜਨਰੇਸ਼ਨ API ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਚਿੱਤਰ ਬਣਾਓ
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣਾ ਪ੍ਰੰਪਟ ਟੈਕਸਟ ਇੱਥੇ ਦਾਖ਼ਲ ਕਰੋ
            size='1024x1024',
            n=1
        )
        # ਸਟੋਰ ਕੀਤੇ ਚਿੱਤਰ ਲਈ ਡਾਇਰੈਕਟਰੀ ਸੈੱਟ ਕਰੋ
        image_dir = os.path.join(os.curdir, 'images')

        # ਜੇ ਡਾਇਰੈਕਟਰੀ ਮੌਜੂਦ ਨਹੀਂ ਹੈ, ਤਾਂ ਇਸ ਨੂੰ ਬਣਾਓ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ਚਿੱਤਰ ਪਾਥ ਸ਼ੁਰੂ ਕਰੋ (ਦਿਆਨ ਦਿਓ ਕਿ ਫਾਈਲ ਟਾਈਪ png ਹੋਣਾ ਚਾਹੀਦਾ ਹੈ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # ਜਨਰੇਟ ਕੀਤਾ ਚਿੱਤਰ ਪ੍ਰਾਪਤ ਕਰੋ
        print(generation_response)

        image_url = generation_response.data[0].url # ਜਵਾਬ ਵਿੱਚੋਂ ਚਿੱਤਰ URL ਕੱਢੋ
        generated_image = requests.get(image_url).content  # ਚਿੱਤਰ ਡਾਊਨਲੋਡ ਕਰੋ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ਡਿਫਾਲਟ ਚਿੱਤਰ ਵਿਊਅਰ ਵਿੱਚ ਚਿੱਤਰ ਦਿਖਾਓ
        image = Image.open(image_path)
        image.show()

    # ਅਪਵਾਦਾਂ ਨੂੰ ਫੜੋ
    except openai.BadRequestError as err:
        print(err)

    ```

ਆਓ ਇਸ ਕੋਡ ਨੂੰ ਸਮਝੀਏ:

- ਪਹਿਲਾਂ, ਅਸੀਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਪੋਰਟ ਕਰਦੇ ਹਾਂ, ਜਿਨ੍ਹਾਂ ਵਿੱਚ OpenAI ਲਾਇਬ੍ਰੇਰੀ, dotenv ਲਾਇਬ੍ਰੇਰੀ, requests ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ Pillow ਲਾਇਬ੍ਰੇਰੀ ਸ਼ਾਮਲ ਹਨ।

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv 
    ```

- ਇਸ ਤੋਂ ਬਾਅਦ, ਅਸੀਂ ਉਹ ਅਬਜੈਕਟ ਬਣਾਉਂਦੇ ਹਾਂ, ਜੋ ਤੁਹਾਡੇ ``.env`` ਤੋਂ API ਕੁੰਜੀ ਪ੍ਰਾਪਤ ਕਰੇਗਾ।

    ```python
        # OpenAI ਓਬਜੈਕਟ ਬਣਾਓ
        client = openai.OpenAI()
    ```

- ਅਗਲਾ ਕੰਮ, ਅਸੀਂ ਚਿੱਤਰ ਤਿਆਰ ਕਰਦੇ ਹਾਂ:

    ```python
    # ਇਮੇਜ ਜੈਨਰੇਸ਼ਨ ਏਪੀਆਈ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਚਿੱਤਰ ਬਣਾਓ
    generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣਾ ਪ੍ਰਾਂਪਟ ਲਗਾਓ ਇੱਥੇ
        size='1024x1024',
        n=1
    )
    ```

    ਉਪਰਲੇ ਕੋਡ ਦਾ ਜਵਾਬ ਇੱਕ JSON ਅਬਜੈਕਟ ਹੈ ਜੋ ਤਿਆਰ ਕੀਤੇ ਚਿੱਤਰ ਦਾ URL ਸਹਿਤ ਹੁੰਦਾ ਹੈ। ਅਸੀਂ ਇਸ URL ਨੂੰ ਚਿੱਤਰ ਡਾਊਨਲੋਡ ਕਰਨ ਅਤੇ ਫਾਈਲ ਵਿੱਚ ਸੇਵ ਕਰਨ ਲਈ ਵਰਤ ਸਕਦੇ ਹਾਂ।

- ਆਖ਼ਿਰ ਵਿੱਚ, ਅਸੀਂ ਚਿੱਤਰ ਖੋਲ੍ਹਦੇ ਹਾਂ ਅਤੇ ਸਮਾਨ ਚਿੱਤਰ ਦਰਸ਼ਕ ਨਾਲ ਇਸਨੂੰ ਪ੍ਰਦਰਸ਼ਿਤ ਕਰਦੇ ਹਾਂ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਬਾਰੇ ਹੋਰ ਵੇਰਵੇ

ਆਓ ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਵਾਲੇ ਕੋਡ ਨੂੰ ਵਧੇਰੇ ਵਿਸਥਾਰ ਨਾਲ ਦੇਖੀਏ:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਇੱਥੇ ਆਪਣਾ ਪ੍ਰੰਪਟ ਟੈਕਸਟ ਦਰਜ ਕਰੋ
        size='1024x1024',
        n=1
    )
```

- **prompt**, ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਲਈ ਵਰਤੀ ਜਾਣ ਵਾਲੀ ਲਿਖਤੀ ਪ੍ਰਿਰੋਸ਼ਨਾ ਹੈ। ਇਸ ਮਾਮਲੇ ਵਿੱਚ, ਅਸੀਂ “Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils” ਪ੍ਰਿਰੋਸ਼ਨਾ ਦਾ ਉਪਯੋਗ ਕਰ ਰਹੇ ਹਾਂ।
- **size**, ਬਣਾਈ ਜਾ ਰਹੀ ਚਿੱਤਰ ਦਾ ਆਕਾਰ ਹੈ। ਇਸ ਮਾਮਲੇ ਵਿੱਚ, ਅਸੀਂ 1024x1024 ਪਿਕਸਲ ਦਾ ਚਿੱਤਰ ਤਿਆਰ ਕਰ ਰਹੇ ਹਾਂ।
- **n**, ਚਿੱਤਰਾਂ ਦੀ ਗਿਣਤੀ ਹੈ ਜੋ ਤਿਆਰ ਕੀਤੀਆਂ ਜਾ ਰਹੀਆਂ ਹਨ। ਇਸ ਮਾਮਲੇ ਵਿੱਚ, ਅਸੀਂ ਦੋ ਚਿੱਤਰ ਤਿਆਰ ਕਰ ਰਹੇ ਹਾਂ।

ਹੋਰ ਵੀ ਚਿੱਤਰਾਂ ਨਾਲ ਕੀ ਕਰ ਸਕਦੇ ਹੋ ਉਸ ਬਾਰੇ ਅਗਲੇ ਸੈਕਸ਼ਨ ਵਿੱਚ ਜ਼ਿਕਰ ਹੋਵੇਗਾ।

## ਚਿੱਤਰ ਤਿਆਰ ਕਰਨ ਦੀ ਵਾਧੂ ਸਮਰੱਥਾਵਾਂ

ਹੁਣ ਤੱਕ ਤੁਸੀਂ ਦੇਖਿਆ ਕਿ ਅਸੀਂ ਸਿਰਫ ਕੁਝ ਕਤਾਰਾਂ ਵਿੱਚ ਪਾਈਥਨ ਦੀ ਵਰਤੋਂ ਨਾਲ ਚਿੱਤਰ ਕਿਵੇਂ ਤਿਆਰ ਕਰ ਸਕਦੇ ਹਾਂ। ਪਰ ਚਿੱਤਰਾਂ ਨਾਲ ਹੋਰ ਵੀ ਕੁਝ ਕੀਤਾ ਜਾ ਸਕਦਾ ਹੈ।

ਤੁਸੀਂ ਇਹ ਵੀ ਕਰ ਸਕਦੇ ਹੋ:

- **ਸੰਸ਼ੋਧਨ ਕਰੋ**। ਮੌਜੂਦਾ ਚਿੱਤਰ ਨੂੰ ਮਾਸਕ ਅਤੇ ਪ੍ਰਿਰੋਸ਼ਨਾ ਦੇ ਕੇ ਤੁਸੀਂ ਚਿੱਤਰ ਨੂੰ ਬਦਲ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਲਈ, ਤੁਸੀਂ ਚਿੱਤਰ ਦੇ ਕਿਸੇ ਖੰਡ ਵਿੱਚ ਕੁਝ ਸ਼ਾਮਲ ਕਰ ਸਕਦੇ ਹੋ। ਸੋਚੋ ਸਾਡਾ ਖਰਗੋਸ਼ ਚਿੱਤਰ ਹੈ, ਤੁਸੀਂ ਖਰਗੋਸ਼ ਨੂੰ ਟੋਪੀ ਪਹਿਨਾ ਸਕਦੇ ਹੋ। ਇਹ ਤਰੀਕਾ ਹੈ ਕਿ ਤੁਸੀਂ ਚਿੱਤਰ, ਮਾਸਕ (ਜਿਸ ਨਾਲ ਬਦਲਾਅ ਵਾਲੇ ਖੇਤਰ ਦੀ ਪਹਿਚਾਣ ਹੁੰਦੀ ਹੈ) ਅਤੇ ਲਿਖਤੀ ਪ੍ਰਿਰੋਸ਼ਨਾ ਜੋ ਦੱਸੇ ਕਿ ਕੀ ਕਰਨਾ ਹੈ, ਦਿੰਦੇ ਹੋ।

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ਮੁੱਢਲਾ ਚਿੱਤਰ ਸਿਰਫ ਖਰਗੋਸ਼ ਨੂੰ ਦਿਖਾਏਗਾ ਪਰ ਆਖ਼ਰੀ ਚਿੱਤਰ ਵਿੱਚ ਖਰਗੋਸ਼ 'ਤੇ ਟੋਪੀ ਹੋਵੇਗੀ।
    
- **ਵੱਖ-ਵੱਖ ਸੰਸਕਰਨ ਬਣਾਓ**। ਇਸ ਵਿਚਾਰ ਦੇ ਨਾਲ ਕਿ ਤੁਸੀਂ ਮੌਜੂਦਾ ਚਿੱਤਰ ਲੈਂਦੇ ਹੋ ਅਤੇ ਕਹਿੰਦੇ ਹੋ ਕਿ ਵੱਖ-ਵੱਖ ਸੰਸਕਰਨ ਬਣਾਏ ਜਾਣ। ਵੱਖ-ਵੱਖ ਸੰਸਕਰਨ ਬਣਾਉਣ ਲਈ, ਤੁਸੀਂ ਚਿੱਤਰ ਅਤੇ ਲਿਖਤੀ ਪ੍ਰਿਰੋਸ਼ਨਾ ਦਿੰਦੇ ਹੋ ਅਤੇ ਇਸੇ ਤਰ੍ਹਾਂ ਕੋਡ ਲਿਖਦੇ ਹੋ:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
