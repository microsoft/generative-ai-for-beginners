# ਇਮેજ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ (ਅਜ਼ੂਰ ਓਪਨਏਆਈ)

ਸਿਰਫ਼ ਲੇਖ ਜਨਰੇਸ਼ਨ ਹੀ ਨਹੀਂ, ਤੁਸੀਂ ਟੈਕਸਟ ਵਰਣਨਾਂ ਤੋਂ ਤਸਵੀਰਾਂ ਵੀ ਬਣਾ ਸਕਦੇ ਹੋ। ਤਸਵੀਰਾਂ ਇੱਕ ਮੋਡੈਲਿਟੀ ਵਜੋਂ ਮੇਡਟੈਕ, ਆਰਕੀਟੈਕਚਰ, ਟੂਰਿਜ਼ਮ, ਗੇਮ ਡਿਵੈਲਪਮੈਂਟ, ਮਾਰਕੀਟਿੰਗ ਅਤੇ ਹੋਰ ਥਾਵਾਂ 'ਤੇ ਲਾਭਦਾਇਕ ਹੁੰਦੀਆਂ ਹਨ। ਇਸ ਪਾਠ ਵਿੱਚ ਅਸੀਂ Microsoft Foundry 'ਤੇ ਅੱਜ ਦੇ **GPT ਇਮેજ** ਮਾਡਲਾਂ ਨਾਲ ਕੰਮ ਕਰਦੇ ਹਾਂ।

## ਸਿੱਖਣ ਦੇ ਲਕੜੀ ਨਿਸ਼ਾਨ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਤੱਕ ਤੁਸੀਂ ਸਕੋਗੇ:

- ਸਮਝਾਓ ਕਿ ਇਮেজ ਜਨਰੇਸ਼ਨ ਕੀ ਹੈ ਅਤੇ ਇਹ ਕਿੱਥੇ ਲਾਭਦਾਇਕ ਹੈ।
- `gpt-image` ਮਾਡਲ ਪਰਿਵਾਰ ਨੂੰ ਸਮਝੋ ਅਤੇ ਇਹ ਪੁਰਾਣੇ DALL·E ਮਾਡਲਾਂ ਤੋਂ ਕਿਵੇਂ ਵੱਖਰਾ ਹੈ।
- ਇੱਕ ਇਮেজ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਓ ਅਤੇ ਤਸਵੀਰਾਂ ਅਦਿੱਠੋ।

## ਇਮેજ ਜਨਰੇਸ਼ਨ ਕੀ ਹੈ?

ਇਮೇಜ ਜਨਰੇਸ਼ਨ ਮਾਡਲ ਟੈਕਸਟ ਪ੍ਰਾਂਪਟ ਤੋਂ ਤਸਵੀਰਾਂ ਬਣਾਉਂਦੇ ਹਨ। ਆਧੁਨਿਕ ਮਾਡਲ ਜਿਵੇਂ ਕਿ `gpt-image` ਟੈਕਸਟ ਅਤੇ ਤਸਵੀਰਾਂ ਦੀ ਸਿੱਖਣ ਰਹਿਤ ਸੰਬੰਧ ਨੂੰ ਟ੍ਰੇਨਿੰਗ ਦੌਰਾਨ ਸਿੱਖਦੇ ਹਨ, ਫਿਰ ਜਾਦੂਈ ਢੰਗ ਨਾਲ ਰੈਂਡਮ ਅਵਾਜ਼ ਨੂੰ ਇੱਕ ਤਸਵੀਰ ਵਿੱਚ ਬਦਲਦੇ ਹਨ ਜੋ ਤੁਹਾਡੇ ਪ੍ਰਾਂਪਟ ਨਾਲ ਮੇਲ ਖਾਂਦੀ ਹੈ।

ਦੋ ਪ੍ਰਸਿਧ ਤਸਵੀਰ ਮਾਡਲ ਪਰਿਵਾਰ ਹਨ:

- **`gpt-image` (OpenAI)** — ਇਸ ਪਾਠ ਵਿੱਚ ਵਰਤੇ ਜਾਣ ਵਾਲੇ ਮੌਜੂਦਾ ਜਨਰੇਸ਼ਨ। ਇਹ ਟੈਕਸਟ-ਟੂ-ਇਮેજ ਜਨਰੇਸ਼ਨ ਅਤੇ ਤਸਵੀਰ ਐਡੀਟਿੰਗ (ਮਾਸਕ ਨਾਲ ਇੰਪੇਂਟਿੰਗ) ਨੂੰ ਸਹਿਯੋਗ ਦਿੰਦਾ ਹੈ।
- **Midjourney** — ਆਪਣੀ ਸਰਵਿਸ ਅਤੇ ਡਿਸਕਾਰਡ-ਅਧਾਰਿਤ ਵਰਕਫਲੋਅ ਵਾਲਾ ਪ੍ਰਸਿੱਧ ਤੀਸਰਾ-ਪੱਖ ਮਾਡਲ।

> ਪੁਰਾਣੇ OpenAI ਇਮਾਜ਼ ਮਾਡਲ — **DALL·E 2** ਅਤੇ **DALL·E 3** — ਪੁਰਾਣੇ ਹਨ। DALL·E 3 ਹੁਣ ਨਵੀਆਂ ਡਿਪਲੋਇਮੈਂਟਾਂ ਲਈ ਉਪਲਬਧ ਨਹੀਂ ਹੈ, ਅਤੇ ਜਿਵੇਂ `create_variation` ਫੀਚਰ ਸਿਰਫ਼ DALL·E 2 ਵਿੱਚ ਸੀ। ਨਵੀਆਂ ਐਪਲੀਕੇਸ਼ਨਾਂ ਲਈ `gpt-image` ਮਾਡਲ ਵਰਤੋ।

Microsoft Foundry 'ਤੇ, **`gpt-image-2`** ਤਾਜ਼ਾ ਅਤੇ ਸਭ ਤੋਂ ਸਮਰਥ ਇਮਾਜ਼ ਮਾਡਲ ਹੈ ਅਤੇ ਇਹ ਸੁਝਾਏ ਗਏ ਡਿਫੌਲਟ ਹੈ। `gpt-image-1.5` ਅਤੇ `gpt-image-1-mini` ਵੀ ਆਮ ਤੌਰ 'ਤੇ ਉਪਲਬਧ ਹਨ।

> **ਮਹਤਵਪੂਰਨ:** `gpt-image` ਮਾਡਲਾਂ ਦੁਆਰਾ ਬਣਾਈ ਗਈ ਤਸਵੀਰ **base64** (`b64_json`) ਵਜੋਂ ਵਾਪਸ ਮਿਲਦੀ ਹੈ, URL ਵਜੋਂ ਨਹੀਂ। ਤੁਹਾਡਾ ਕੋਡ base64 ਸਟਰਿੰਗ ਨੂੰ ਬਾਈਟਸ ਵਿੱਚ ਡੀਕੋਡ ਕਰਦਾ ਹੈ ਅਤੇ ਸੇਵ ਕਰਦਾ ਹੈ — ਕੋਈ ਤਸਵੀਰ URL ਡਾਊਨਲੋਡ ਕਰਨ ਲਈ ਨਹੀਂ ਹੁੰਦਾ।


## ਆਪਣੀ ਪਹਿਲੀ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ

ਤਾਂ ਇੱਕ ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣ ਲਈ ਕੀ ਲੋੜ ਹੈ? ਤੁਹਾਨੂੰ ਹੇਠ ਲਿਖੀਆਂ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਦੀ ਲੋੜ ਹੈ:

- **python-dotenv**, ਤੁਹਾਨੂੰ ਇਸ ਲਾਇਬ੍ਰੇਰੀ ਨੂੰ ਆਪਣੀਆਂ ਗੁਪਤ ਜਾਣਕਾਰੀਆਂ ਨੂੰ ਕੋਡ ਤੋਂ ਦੂਰ *.env* ਫਾਇਲ ਵਿੱਚ ਰੱਖਣ ਲਈ ਬਹੁਤ ਸਿਫਾਰਸ਼ ਕੀਤੀ ਜਾਂਦੀ ਹੈ।
- **openai**, ਇਸ ਲਾਇਬ੍ਰੇਰੀ ਨੂੰ ਤੁਸੀਂ OpenAI API ਨਾਲ ਸੰਵਾਦ ਕਰਨ ਲਈ ਵਰਤੋਂਗੇ।
- **pillow**, ਪਾਇਥਨ ਵਿੱਚ ਚਿੱਤਰਾਂ ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ।

ਜੇ ਪਹਿਲਾਂ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ਪੰਨੇ 'ਤੇ ਦਿੱਤੇ ਨਿਰਦੇਸ਼ਾਂ ਦੀ ਪਾਲਣਾ ਕਰਦਿਆਂ ਇੱਕ Microsoft Foundry ਸਾਧਨ ਅਤੇ ਮਾਡਲ ਬਣਾਓ। ਮਾਡਲ ਵਜੋਂ **gpt-image-2** ਚੁਣੋ (ਤਾਜ਼ਾ Azure OpenAI ਚਿੱਤਰ ਮਾਡਲ; DALL·E ਲੇਗਸੀ ਹੈ)।

1. ਹੇਠ ਲਿਖੇ ਸਮੱਗਰੀ ਨਾਲ ਇੱਕ *.env* ਫਾਇਲ ਬਣਾਓ:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    ਇਸ ਜਾਣਕਾਰੀ ਨੂੰ ਆਪਣੇ ਸਾਧਨ ਲਈ Microsoft Foundry ਪੋਰਟਲ ਦੇ "Deployments" ਸੈਕਸ਼ਨ ਵਿੱਚ ਲੱਭੋ।



1. ਉਪਰੋਕਤ ਲਾਇਬ੍ਰੇਰੀਆਂ ਨੂੰ ਇੱਕ ਫਾਈਲ ਵਿੱਚ ਇੱਕੱਠਾ ਕਰੋ ਜਿਸਦਾ ਨਾਮ *requirements.txt* ਰੱਖੋ ਇਸ ਤਰ੍ਹਾਂ:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. ਅਗਲੇ, ਵਰਚੂਅਲ ਵਾਤਾਵਰਨ ਬਣਾਓ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਸਟਾਲ ਕਰੋ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> ਵਿੰਡੋਜ਼ ਲਈ, ਆਪਣਾ ਵਿਰਚੁਅਲ ਵਾਤਾਵਰਣ ਬਣਾਉਣ ਅਤੇ ਐਕਟੀਵੇਟ ਕਰਨ ਲਈ ਹੇਠ ਲਿਖੇ ਕਮਾਂਡਾਂ ਦੀ ਵਰਤੋਂ ਕਰੋ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. ਫਾਈਲ *app.py* ਵਿੱਚ ਹੇਠ ਲਿਖਿਆ ਕੋਡ ਸ਼ਾਮਲ ਕਰੋ:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv ਆਯਾਤ ਕਰੋ
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) ਕਲਾਇੰਟ ਕਾਨਫਿਗਰ ਕਰੋ।
    # ਚਿੱਤਰ ਮਾਡਲਾਂ ਨੂੰ ਇੱਕ ਹਾਲੀਆ API ਸੰਸਕਰਣ ਦੀ ਲੋੜ ਹੁੰਦੀ ਹੈ — ਆਪਣੀ ਮਾਡਲ ਲਈ Foundry ਦਸਤਾਵੇਜ਼ਾਂ ਚੈੱਕ ਕਰੋ।
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # ਚਿੱਤਰ ਜਨਰੇਸ਼ਨ API ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਚਿੱਤਰ ਬਣਾਓ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਇੱਥੇ ਆਪਣਾ ਪ੍ਰਾਰੰਭਿਕ ਟੈਕਸਟ ਦਾਖਲ ਕਰੋ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # ਉਦਾਹਰਨ ਵਜੋਂ "gpt-image-2"
        )
        # ਸਟੋਰ ਕੀਤੇ ਚਿੱਤਰ ਲਈ ਡਾਇਰੈਕਟਰੀ ਸੈੱਟ ਕਰੋ
        image_dir = os.path.join(os.curdir, 'images')

        # ਜੇ ਡਾਇਰੈਕਟਰੀ ਮੌਜੂਦ ਨਾ ਹੋਵੇ, ਤਾਂ ਇਹ ਬਣਾਓ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ਚਿੱਤਰ ਦਾ ਰਸਤਾ ਸ਼ੁਰੂ ਕਰੋ (ਨੋਟ ਕਰੋ ਕਿ ਫਾਇਲ ਦਾ ਪ੍ਰਕਾਰ png ਹੋਣਾ ਚਾਹੀਦਾ ਹੈ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image ਮਾਡਲ ਚਿੱਤਰ ਨੂੰ base64 (b64_json) ਵਜੋਂ ਵਾਪਸ ਕਰਦੇ ਹਨ, URL ਨਹੀਂ
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ਡਿਫੌਲਟ ਚਿੱਤਰ ਦੇਖਣ ਵਾਲੇ ਵਿੱਚ ਚਿੱਤਰ ਦਿਖਾਓ
        image = Image.open(image_path)
        image.show()

    # исключения ਨੂੰ ਫੜੋ
    except BadRequestError as err:
        print(err)

    ```

ਆਓ ਇਸ ਕੋਡ ਦੀ ਵਿਆਖਿਆ ਕਰੀਏ:

- ਸਭ ਤੋਂ ਪਹਿਲਾਂ, ਅਸੀਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਇੰਮਪੋਰਟ ਕਰਦੇ ਹਾਂ ਜਿਨ੍ਹਾਂ ਦੀ ਸਾਨੂੰ ਲੋੜ ਹੈ, ਜਿਸ ਵਿੱਚ OpenAI ਲਾਇਬ੍ਰੇਰੀ, dotenv ਲਾਇਬ੍ਰੇਰੀ, base64 ਮੋਡੀਯੂਲ ਅਤੇ Pillow ਲਾਇਬ੍ਰੇਰੀ ਸ਼ਾਮਲ ਹਨ।

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- ਅਗਲੇ, ਅਸੀਂ *.env* ਫਾਈਲ ਤੋਂ ਐਨਵਾਇਰਨਮੈਂਟ ਵੈਰੀਏਬਲ ਲੋਡ ਕਰਦੇ ਹਾਂ।

    ```python
    # ਡੌਟਏਨਵ ਨੂੰ ਆਯਾਤ ਕਰੋ
    dotenv.load_dotenv()
    ```

- ਇਸ ਤੋਂ ਬਾਅਦ, ਅਸੀਂ Azure OpenAI (Microsoft Foundry) ਕਲਾਇੰਟ ਨੂੰ ਸੰਰਚਿਤ ਕਰਦੇ ਹਾਂ।

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- ਅਗਲਾ, ਅਸੀਂ ਚਿੱਤਰ ਤਿਆਰ ਕਰਦੇ ਹਾਂ:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣਾ ਪ੍ਰੰਪਟ ਟੈਕਸਟ ਇੱਥੇ ਦਰਜ ਕਰੋ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` ਮਾਡਲ ਤਸਵੀਰ ਨੂੰ **base64** ਸਤਰ ਵਜੋਂ `data[0].b64_json` ਵਿੱਚ ਵਾਪਸ ਕਰਦਾ ਹੈ। ਅਸੀਂ ਇਸ ਨੂੰ ਬਾਈਟਸ ਵਿੱਚ ਡਿਕੋਡ ਕਰਦੇ ਹਾਂ ਅਤੇ ਫਾਈਲ ਵਿੱਚ ਲਿਖਦੇ ਹਾਂ — ਡਾਊਨਲੋਡ ਕਰਨ ਲਈ ਕੋਈ URL ਨਹੀਂ ਹੁੰਦਾ।

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- ਆਖਿਰਕਾਰ, ਅਸੀਂ ਤਸਵੀਰ ਖੋਲ੍ਹਦੇ ਹਾਂ ਅਤੇ ਮਾਨਕ ਤਸਵੀਰ ਵੇਖਣ ਵਾਲਾ ਵਰਤ ਕੇ ਇਸਨੂੰ ਦਿਖਾਉਂਦੇ ਹਾਂ:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ਤਸਵੀਰ ਬਣਾਉਣ ਬਾਰੇ ਹੋਰ ਵੇਰਵੇ

ਆਓ `images.generate` ਦੇ ਪੈਰਾਮੀਟਰਾਂ ਨੂੰ ਵੇਖੀਏ:

- **prompt**, ਉਹ ਟੈਕਸਟ ਪ੍ਰਾਮਪਟ ਹੈ ਜੋ ਤਸਵੀਰ ਬਣਾਉਣ ਲਈ ਵਰਤਿਆ ਜਾਂਦਾ ਹੈ। ਇੱਥੇ ਇਹ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" ਹੈ।
- **size**, ਬਣਾਈ ਗਈ ਤਸਵੀਰ ਦਾ ਆਕਾਰ ਹੈ (ਉਦਾਹਰਨ ਲਈ `1024x1024`, `1536x1024`, `1024x1536`, ਜਾਂ `"auto"`).
- **n**, ਤਸਵੀਰਾਂ ਦੀ ਗਿਣਤੀ ਹੈ। ਇੱਥੇ ਅਸੀਂ ਇੱਕ ਤਸਵੀਰ ਬਣਾਈ ਹੈ।
- **model**, ਤੁਹਾਡੇ ਚਿੱਤਰ ਮਾਡਲ ਡਿਪਲੋਇਮੈਂਟ ਦਾ ਨਾਮ ਹੈ (ਜਿਵੇਂ `gpt-image-2`)।

> ਚਿੱਤਰ ਮਾਡਲ `temperature` ਪੈਰਾਮੀਟਰ ਨਹੀਂ ਲੈਂਦੇ — ਇਹ ਇੱਕ ਟੈਕਸਟ-ਬਣਾਉਣ ਕন্ট੍ਰੋਲ ਹੈ। ਵੱਖ-ਵੱਖ ਪ੍ਰਕਾਰ ਦੀਆਂ ਚੀਜ਼ਾਂ ਲਈ, API ਨੂੰ ਮੁੜ ਕਾਲ ਕਰੋ; ਘਟਾਉਣ ਲਈ, ਆਪਣਾ ਪ੍ਰਾਮਪਟ ਹੋਰ ਵਿਸ਼ੇਸ਼ ਬਣਾਓ।

## ਚਿੱਤਰ ਬਣਾਉਣ ਦੀ ਵਾਧੂ ਸਮਰੱਥਾ

ਤੁਸੀਂ ਦੇਖਿਆ ਕਿ ਕਿਵੇਂ ਕੁਝ ਲਾਈਨਾਂ ਦੇ Python ਨਾਲ ਤਸਵੀਰ ਬਣਾਈ ਜਾ ਸਕਦੀ ਹੈ। `gpt-image` ਮਾਡਲ ਮੌਜੂਦਾ ਤਸਵੀਰ ਨੂੰ **ਸੰਪਾਦਿਤ** ਵੀ ਕਰ ਸਕਦੇ ਹਨ। ਇੱਕ ਤਸਵੀਰ, ਇਕ ਵਿਕਲਪੀ **ਮਾਸਕ** (ਜੋ ਬਦਲਾਅ ਲਈ ਖੇਤਰ ਨੂੰ ਦਰਸਾਉਂਦਾ ਹੈ), ਅਤੇ ਇੱਕ ਪ੍ਰਾਮਪਟ ਦੇ ਕੇ ਤੁਸੀਂ ਤਸਵੀਰ ਦਾ ਕੋਈ ਹਿੱਸਾ ਬਦਲ ਸਕਦੇ ਹੋ — ਉਦਾਹਰਨ ਵਜੋਂ, ਸਾਡੇ ਖਰਗੋਸ਼ ਨੂੰ ਇੱਕ ਟੋਪੀ ਪਹਿਨਾਉਣਾ।

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ਸੰਸ਼ੋਧਨ ਵੀ ਬੇਸ64 ਵਜੋਂ ਵਾਪਸ ਕੀਤੇ ਜਾਂਦੇ ਹਨ
edited_image = base64.b64decode(response.data[0].b64_json)
```
