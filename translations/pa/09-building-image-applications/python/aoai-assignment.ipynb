## ਆਪਣੀ ਪਹਿਲੀ ਇਮੇਜ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣਾ

ਤਾਂ ਫਿਰ ਇਮੇਜ ਜਨਰੇਸ਼ਨ ਐਪਲੀਕੇਸ਼ਨ ਬਣਾਉਣ ਲਈ ਕੀ ਚਾਹੀਦਾ ਹੈ? ਤੁਹਾਨੂੰ ਹੇਠ ਲਿਖੀਆਂ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਦੀ ਲੋੜ ਹੈ:

- **python-dotenv**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਸੀਕ੍ਰੇਟਸ ਨੂੰ ਕੋਡ ਤੋਂ ਦੂਰ *.env* ਫਾਇਲ ਵਿੱਚ ਰੱਖਣ ਲਈ ਬਹੁਤ ਸਿਫ਼ਾਰਸ਼ੀ ਹੈ।
- **openai**, ਇਸ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਤੁਸੀਂ OpenAI API ਨਾਲ ਇੰਟਰਐਕਟ ਕਰਨ ਲਈ ਕਰੋਗੇ।
- **pillow**, ਪਾਇਥਨ ਵਿੱਚ ਚਿੱਤ੍ਰਾਂ ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ।
- **requests**, HTTP ਬੇਨਤੀਆਂ ਕਰਨ ਵਿੱਚ ਮਦਦ ਲਈ।

ਜੇ ਪਹਿਲਾਂ ਨਾ ਕੀਤਾ ਹੋਵੇ, ਤਾਂ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ਪੰਨੇ ਉੱਤੇ ਦਿੱਤੀਆਂ ਹਦਾਇਤਾਂ ਦੀ ਪਾਲਨਾ ਕਰਦੇ ਹੋਏ ਇੱਕ Azure OpenAI ਰਿਸੋਰਸ ਅਤੇ ਮਾਡਲ ਬਣਾਓ। ਮਾਡਲ ਵਜੋਂ **gpt-image-1** ਚੁਣੋ (ਮੌਜੂਦਾ ਜਨਰੇਸ਼ਨ Azure OpenAI ਇਮੇਜ ਮਾਡਲ; DALL-E 3 ਲੇਗੇਸੀ ਹੈ ਅਤੇ ਨਵੇਂ ਡਿਪਲੋਇਮੈਂਟ ਲਈ ਹੋਰ ਉਪਲਬਧ ਨਹੀਂ ਹੈ)।

1. *.env* ਨਾਮਕ ਇਕ ਫਾਇਲ ਬਣਾਓ ਜਿਸ ਵਿੱਚ ਹੇਠ ਲਿਖਿਆ ਸਮੱਗਰੀ ਹੋਵੇ:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    ਇਸ ਜਾਣਕਾਰੀ ਨੂੰ ਤੇਰੇ ਰਿਸੋਰਸ ਲਈ Microsoft Foundry ਪੋਰਟਲ ਵਿੱਚ "Deployments" ਸੈਕਸ਼ਨ ਵਿੱਚ ਲੱਭੋ।



1. ਉਪਰੋਕਤ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਨੂੰ *requirements.txt* ਨਾਮਕ ਫਾਇਲ ਵਿੱਚ ਇਸ ਤਰ੍ਹਾਂ ਇਕੱਤਰ ਕਰੋ:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. ਅਗਲਾਂ, ਵਰਚੁਅਲ ਵਾਤਾਵਰਣ ਬਣਾਓ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਇੰਸਟਾਲ ਕਰੋ:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> ਵਿੰਡੋਜ਼ ਲਈ, ਆਪਣਾ ਵਰਚੁਅਲ ਵਾਤਾਵਰਨ ਬਣਾਉਣ ਅਤੇ ਐਕਟੀਵੇਟ ਕਰਨ ਲਈ ਹੇਠਾਂ ਦਿੱਤੇ ਕਮਾਂਡਾਂ ਦੀ ਵਰਤੋਂ ਕਰੋ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. ਫਾਈਲ *app.py* ਵਿੱਚ ਹੇਠਾਂ ਦਿੱਤਾ ਕੋਡ ਸ਼ਾਮਲ ਕਰੋ:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv ਨੂੰ ਇੰਪੋਰਟ ਕਰੋ
    dotenv.load_dotenv()

    # Azure OpenAI ਸੇਵਾ ਕਲਾਇੰਟ ਨੂੰ ਕਨਫਿਗਰ ਕਰੋ
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # ਚਿੱਤ੍ਰ ਸਿਰਜਣ API ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਚਿੱਤ੍ਰ ਬਣਾਓ
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ਆਪਣਾ ਪ੍ਰਾਂਪਟ ਟੈਕਸਟ ਇੱਥੇ ਦਰਜ ਕਰੋ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # ਸਟੋਰ ਕੀਤੇ ਚਿੱਤ੍ਰ ਲਈ ਡਾਇਰੈਕਟਰੀ ਸੈੱਟ ਕਰੋ
        image_dir = os.path.join(os.curdir, 'images')
    
        # ਜੇ ਡਾਇਰੈਕਟਰੀ ਮੌਜੂਦ ਨਹੀਂ ਹੈ ਤਾਂ ਉਸ ਨੂੰ ਬਣਾਓ
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # ਚਿੱਤ੍ਰ ਪੱਥ ਸ਼ੁਰੂ ਕਰੋ (ਨੋਟ: ਫਾਈਲ ਟਾਈਪ PNG ਹੋਣਾ ਚਾਹੀਦਾ ਹੈ)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # ਬਣਾਇਆ ਗਿਆ ਚਿੱਤ੍ਰ ਪ੍ਰਾਪਤ ਕਰੋ
        image_url = generation_response.data[0].url  # ਜਵਾਬ ਤੋਂ ਚਿੱਤ੍ਰ URL ਨਿਕਾਲੋ
        generated_image = requests.get(image_url).content  # ਚਿੱਤ੍ਰ ਡਾਊਨਲੋਡ ਕਰੋ
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # ਡਿਫਾਲਟ ਚਿੱਤ੍ਰ ਦਰਸ਼ਕ ਵਿੱਚ ਚਿੱਤ੍ਰ ਦਿਖਾਓ
        image = Image.open(image_path)
        image.show()
    
    # ਵਿਸ਼ੇਸ਼ਤਾਵਾਂ ਨੂੰ ਕੈਚ ਕਰੋ
    except openai.BadRequestError as err:
        print(err)

    ```
