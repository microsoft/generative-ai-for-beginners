# ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਪਾਠਕ੍ਰਮ ਵਿਚ ਦੇਖਿਆ ਹੈ ਕਿ ਮੁੱਖ ਧਾਰਾ ਦੇ ਸੰਕਲਪ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰੋਂਪਟਸ ਅਤੇ ਇੱਥੇ ਤੱਕ ਕਿ ਇੱਕ ਪੂਰੀ ਵਿਭਾਗ ਹੈ ਜਿਸਨੂੰ "ਪ੍ਰੋਂਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਕਹਿੰਦੇ ਹਨ। ਬਹੁਤ ਸਾਰੇ ਸੰਦ ਜਿਨ੍ਹਾਂ ਨਾਲ ਤੁਸੀਂ ਗੱਲ ਕਰ ਸਕਦੇ ਹੋ ਜਿਵੇਂ ChatGPT, Office 365, Microsoft Power Platform ਅਤੇ ਹੋਰ ਬੜੀ ਮਦਦ ਨਾਲ ਤੁਹਾਨੂੰ ਪ੍ਰੋਂਪਟ ਦੀ ਵਰਤੋਂ ਕਰਨ ਵਿੱਚ ਸਹਾਇਤਾ ਕਰਦੇ ਹਨ ਕਿ ਤੁਸੀਂ ਕੁਝ ਪ੍ਰਾਪਤ ਕਰੋ।

ਤੁਹਾਨੂੰ ਅਜਿਹੇ ਤਜ਼ਰਬੇ ਨੂੰ ਇੱਕ ਐਪ ਵਿੱਚ ਸ਼ਾਮਲ ਕਰਨ ਲਈ, ਪ੍ਰੋਂਪਟਸ, ਪੂਰਨਾਮਾਂ ਸਮਝਣੇ ਅਤੇ ਵਰਤੋਂ ਲਈ ਇਕ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣੀਦੀ ਹੈ। ਇਹੀ ਕੁਝ ਤੁਸੀਂ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖੋਗੇ।

## ਪਰੀਚਯ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਉਸਦੇ ਮੁੱਖ ਧਾਰਾਵਾਂ ਬਾਰੇ ਜਾਣੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਵੋਗੇ।
- ਪ੍ਰੋਂਪਟ, ਟੇਮਪਰੇਚਰ, ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰ ਕੇ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਬਣਾਈਏ, ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ 'ਤੇ, ਤੁਸੀਂ ਸਮਰੱਥ ਹੋਵੋਗੇ:

- ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦਾ ਹੈ, ਸਪਸ਼ਟ ਕਰ ਸਕਦੇ ਹੋ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਵੋਗੇ।
- ਤੁਹਾਡੇ ਐਪ ਨੂੰ ਵੱਧ ਜਾਂ ਘੱਟ ਟੋਕਨ ਵਰਤਣ ਲਈ, ਅਤੇ ਟੇਮਪਰੇਚਰ ਨੂੰ ਬਦਲਣ ਲਈ ਸੰਰਚਨਾ ਕਰ ਸਕਦੇ ਹੋ, ਜਿਸ ਨਾਲ ਨਤੀਜਾ ਵੱਖ-ਵੱਖ ਹੋਵੇਗਾ।

## ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਇੱਕ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਉਸਦਾ ਕੋਈ ਨਾ ਕੋਈ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ ਜਿਵੇਂ:

- ਕਮਾਂਡ-ਆਧਾਰਿਤ। ਕਨਸੋਲ ਐਪ ਬੜੀ ਸੱਥਰੀ ਐਪ ਹੁੰਦੇ ਹਨ ਜਿੱਥੋਂ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ ਤੇ ਇਹ ਕਾਰਜ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਲਈ, `git` ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUI) ਰੱਖਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ, ਟੈਕਸਟ ਦਾਖਲ ਕਰਨਾ, ਚੋਣ ਕਰਨਾ ਆਦਿ ਕਰਦੇ ਹੋ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪ ਸੀਮਿਤ ਹੁੰਦੇ ਹਨ

ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਨਾਲ ਤੁਲਨਾ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ:

- **ਇਹ ਸੀਮਿਤ ਹੈ**। ਤੁਸੀਂ ਕਿਸੇ ਵੀ ਕਮਾਂਡ ਨੂੰ ਨਹੀਂ ਟਾਈਪ ਕਰ ਸਕਦੇ, ਸਿਰਫ ਉਹ ਕਮਾਂਡ ਜੋ ਐਪ ਸਮਰਥਨ ਕਰਦਾ ਹੈ।
- **ਭਾਸ਼ਾ-ਵਿਸ਼ੇਸ਼**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਮਰਥਨ ਕਰਦੇ ਹਨ, ਪਰ ਮੂਲ ਤੌਰ 'ਤੇ ਐਪ ਕਿਸੇ ਨਿਰਧਾਰਤ ਭਾਸ਼ਾ ਲਈ ਬਣਾਇਆ ਜਾਂਦਾ ਹੈ, ਚਾਹੇ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਦਾ ਸਮਰਥਨ ਸ਼ਾਮਲ ਕਰ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਦੇ ਲਾਭ

ਤਾਂ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੈ?

ਇੱਕ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ, ਤੁਹਾਡੇ ਕੋਲ ਵਧੇਰੇ ਲਚੀਲਾਪਣ ਹੁੰਦਾ ਹੈ, ਤੁਸੀਂ ਕਮਾਂਡਾਂ ਜਾਂ ਕਿਸੇ ਵਿਸ਼ੇਸ਼ ਦਾਖਲਾ ਭਾਸ਼ਾ ਦੇ ਬੱਧ ਨਹੀਂ ਹੋ। ਇਸ ਦੀ ਥਾਂ, ਤੁਸੀਂ ਪ੍ਰਕ੍ਰਿਤਿਕ ਭਾਸ਼ਾ ਵਿੱਚ ਐਪ ਨਾਲ ਗੱਲ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਲਾਭ ਇਹ ਹੈ ਕਿ ਜਦੋਂ ਤੁਸੀਂ ਇੱਕ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਇੰਟਰੈਕਟ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੇ ਜਾਣਕਾਰੀ ਕਲੈਕਸ਼ਨ 'ਤੇ ਟ੍ਰੇਨ ਕੀਤਾ ਗਿਆ ਹੈ, ਜਦਕਿ ਇੱਕ ਪਾਰੰਪਰਿਕ ਐਪ ਇਕ ਡੇਟਾਬੇਸ ਵਿੱਚ ਸਿਰਫ ਸੀਮਤ ਜਾਣਕਾਰੀ ਤੇ ਕੰਮ ਕਰ ਸਕਦਾ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂ ਸਕਦਾ ਹਾਂ?

ਤੁਸੀਂ ਬਹੁਤ ਕੁਝ ਬਣਾ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਲਈ:

- **ਇੱਕ ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਤੁਹਾਡੇ ਕੰਪਨੀ ਅਤੇ ਉਸਦੇ ਉਤਪਾਦਾਂ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦਿੰਦਾ ਹੈ, ਚੰਗਾ ਵਿਕਲਪ ਹੋ ਸਕਦਾ ਹੈ।
- **ਮਦਦਗਾਰ**। LLMs ਟੈਕਸਟ ਨੂੰ ਸੰਖੇਪ ਕਰਨ, ਜਾਣਕਾਰੀਆਂ ਪ੍ਰਾਪਤ ਕਰਨ, ਰੈਜ਼ਿਊਮੇ ਅਤੇ ਹੋਰ ਟੈਕਸਟ ਬਣਾਉਣ ਵਿੱਚ ਬਹੁਤ ਅਚ্ছে ਹਨ।
- **ਕੋਡ ਅਸਿਸਟੈਂਟ**। ਬੋਲੀ ਮਾਡਲ ਤੇ ਨਿਰਭਰ ਕਰਕੇ, ਤੁਸੀਂ ਇਕ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦਗਾਰ ਕੋਡ ਅਸਿਸਟੈਂਟ ਬਣਾਉਂ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਲਈ, ਤੁਸੀਂ GitHub Copilot ਅਤੇ ChatGPT ਵਰਗੀਆਂ ਚੀਜ਼ਾਂ ਨੂੰ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਲਈ ਵਰਤ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਚੰਗਾ, ਤੁਹਾਨੂੰ LLM ਨਾਲ ਜੁੜਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪਵੇਗਾ ਜੋ ਆਮ ਤੌਰ 'ਤੇ ਦੋ ਤਰੀਕਿਆਂ ਵਿੱਚੋਂ ਇੱਕ ਹੁੰਦਾ ਹੈ:

- API ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣੇ ਪ੍ਰੋਂਪਟ ਨਾਲ ਵੈੱਬ ਬੇਨਤੀਆਂ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਜੈਨਰੇਟ ਕੀਤੇ ਟੈਕਸਟ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਾਂ ਨੂੰ ਸਮੇਤ ਕੇ ਉਨ੍ਹਾਂ ਦੀ ਵਰਤੋਂ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀਆਂ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀਆਂ/SDKs

LLMs ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਜਾਣੇ ਮाने ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਮਾਡਲ ਨਾਲ ਸਾਂਝਾ ਕਰਨ ਅਤੇ ਪ੍ਰੋਂਪਟ ਭੇਜਣ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀ ਹੈ।

ਫਿਰ ਉੱਚ ਸਤਰ 'ਤੇ ਕੰਮ ਕਰਨ ਵਾਲੀਆਂ ਕੁਝ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **Langchain**। Langchain ਵਧੀਆ ਜਾਣੀ ਜਾਂਦੀ ਹੈ ਅਤੇ ਪਾਇਥਨ ਦਾ ਸਮਰਥਨ ਕਰਦੀ ਹੈ।
- **Semantic Kernel**। Semantic Kernel ਮਾਇਕ੍ਰੋਸੌਫਟ ਦੀ ਇਕ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python, ਅਤੇ Java ਫ਼ਰੈਮਵਰਕਾਂ ਦਾ ਸਮਰਥਨ ਕਰਦੀ ਹੈ।

## ChatCompletions ਨਾਲ ਟੈਕਸਟ ਜੈਨਰੇਟ ਕਰੋ

ਟੈਕਸਟ ਬਣਾਉਣ ਦਾ ਤਰੀਕਾ `ChatCompletionsClient` ਕਲਾਸ ਦੀ ਵਰਤੋਂ ਕਰਨਾ ਹੈ।
`samples/python/azure_ai_inference/basic.py` ਵਿੱਚ, ਕੋਡ ਦੇ ਜਵਾਬ ਹਿੱਸੇ ਵਿੱਚ, ਵਪਾਰੀ ਭੂਮਿਕਾ ਲਈ ਕੁਨਟੈਂਟ ਪੈਰਾਮੀਟਰ ਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਤਰੀਕੇ ਨਾਲ ਬਦਲੋ:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ਵਿਕਲਪਿਕ ਪੈਰਾਮੀਟਰ
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **ਤਾਪਮਾਨ ਨਾਲ ਪ੍ਰਯੋਗ**। ਤਾਪਮਾਨ ਇੱਕ ਅਜਿਹੀ ਚੀਜ਼ ਹੈ ਜਿਸਦਾ ਅਸੀਂ ਹੁਣ ਤੱਕ ਜ਼ਿਕਰ ਨਹੀਂ ਕੀਤਾ ਪਰ ਇਹ ਸਾਡੇ ਪ੍ਰੋਗਰਾਮ ਦੇ ਕੰਮ ਕਰਨ ਦੇ ਢੰਗ ਲਈ ਜਰੂਰੀ ਸੰਦਰਭ ਹੈ। ਜੇ ਤਾਪਮਾਨ ਵਧਦਾ ਹੈ ਤਾਂ ਨਤੀਜਾ ਵੱਧ ਬੇਤਹਾਸ਼ਾ ਹੋਵੇਗਾ। ਵਿਰੁੱਧ, ਜੇ ਤਾਪਮਾਨ ਘਟਦਾ ਹੈ ਤਾਂ ਨਤੀਜਾ ਵੱਧ ਅਨੁਮਾਨਯੋਗ ਹੁੰਦਾ ਹੈ। ਸੋਚੋ ਕਿ ਤੁਸੀਂ ਆਪਣੇ ਨਤੀਜੇ ਵਿੱਚ ਵੱਖ-ਵੱਖਪਨ ਚਾਹੁੰਦੇ ਹੋ ਜਾਂ ਨਹੀਂ।

   ਤਾਪਮਾਨ ਬਦਲਣ ਲਈ, ਤੁਸੀਂ `temperature` ਪੈਰਾਮੀਟਰ ਦੀ ਵਰਤੋਂ ਕਰ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਵਜੋਂ, ਜੇ ਤੁਸੀਂ 0.5 ਤਾਪਮਾਨ ਵਰਤਣਾ ਹੈ ਤਾਂ ਤੁਸੀਂ ਐਸਾ ਕਰੋਗੇ:

## ਅਸਾਈਨਮੈਂਟ

ਕਈ ਸੁਝਾਅ ਇਹ ਰਹੇ:

- ਵਿਧੀ ਜਨਰੇਟਰ ਐਪ ਨੂੰ ਹੋਰ ਸੁਧਾਰਨ ਲਈ ਸੋਧ ਕਰੋ। ਤਾਪਮਾਨ ਮੁੱਲਾਂ ਨਾਲ ਅਤੇ ਪ੍ਰੰਪਟਾਂ ਨਾਲ ਖੇਡੋ ਅਤੇ ਦੇਖੋ ਕਿ ਕੀ ਤਿਆਰ ਕਰ ਸਕਦੇ ਹੋ।
- ਇੱਕ "ਸਟਡੀ ਬੱਡੀ" ਬਣਾਓ। ਇਹ ਐਪ ਕਿਸੇ ਮੋਜੂਏ ਤੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇ ਸਕਦੀ ਹੈ, ਉਦਾਹਰਨ ਲਈ ਪਾਇਥਨ, ਤੁਸੀਂ ਪ੍ਰੰਪਟ ਰੱਖ ਸਕਦੇ ਹੋ "ਪਾਇਥਨ ਵਿੱਚ ਕੋਈ ਖਾਸ ਮੋਜੂਾ ਕੀ ਹੈ?" ਜਾਂ ਕੋਈ ਪ੍ਰੰਪਟ ਜਿਸ ਵਿੱਚ ਕੋਡ ਦਿਖਾਉ।
