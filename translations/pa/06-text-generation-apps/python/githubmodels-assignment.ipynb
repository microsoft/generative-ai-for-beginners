# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਕੋਰਸ ਵਿਚ ਦੇਖਿਆ ਹੈ ਕਿ ਕੋਰ ਸਮਝਾਵਟਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰਾਪੰਪਟਸ ਅਤੇ ਇੱਕ ਪੂਰਾ ਵਿਭਾਗ "ਪ੍ਰਾਪੰਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਵੀ ਹੈ। ਕਈ ਟੂਲ ਜਿਵੇਂ ਕਿ ChatGPT, Office 365, Microsoft Power Platform ਅਤੇ ਹੋਰ ਤੁਸੀਂ ਪ੍ਰਾਪੰਪਟਸ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਕੁਝ ਮਕਸਦ ਹਾਸਲ ਕਰਨ ਵਿੱਚ ਸਹਾਇਤਾ ਕਰਦੇ ਹਨ।

ਤੁਹਾਡੇ ਲਈ ਇੱਕ ਐਪ ਵਿੱਚ ਇਹ ਤਜਰਬਾ ਸ਼ਾਮਿਲ ਕਰਨ ਲਈ, ਤੁਹਾਨੂੰ ਪ੍ਰਾਪੰਪਟਸ, ਪੂਰਨਤਾ ਅਤੇ ਕੰਮ ਕਰਨ ਲਈ ਕੋਈ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣਨ ਬਾਰੇ ਸਮਝਣਾ ਹੋਵੇਗਾ। ਇਹੀ ਤੁਹਾਨੂੰ ਇਸ ਚੈਪਟਰ ਵਿੱਚ ਸਿੱਖਣ ਨੂੰ ਮਿਲੇਗਾ।

## ਜਾਣ-ਪਹਚਾਣ

ਇਸ ਅਧਿਆਏ ਵਿੱਚ ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸਦੇ ਮੁੱਖ ਸੰਕਲਪਾਂ ਬਾਰੇ ਸਿੱਖੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓਗੇ।
- ਸਮਝੋਗੇ ਕਿ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣ ਲਈ ਪ੍ਰਾਪੰਪਟ, ਤਾਪਮਾਨ, ਅਤੇ ਟੋਕਨ ਵਰਗੇ ਸੰਕਲਪਾਂ ਕਿਵੇਂ ਵਰਤਣੇ ਹਨ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਵਿੱਚ, ਤੁਸੀਂ ਸਕੋਗੇ:

- ਸਮਝਾਓ ਕਿ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ।
- ਆਪਣੇ ਐਪ ਨੂੰ ਵਧੇਰੇ ਜਾਂ ਘੱਟ ਟੋਕਨ ਦੀ ਵਰਤੋਂ ਅਤੇ ਤਾਪਮਾਨ ਬਦਲਣ ਲਈ ਸੰਰਚਿਤ ਕਰੋ, ਜਿਸ ਨਾਲ ਵੱਖਰੇ ਨਤੀਜੇ ਮਿਲ ਸਕਣ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਕੋਈ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਇਸ ਵਿੱਚ ਕੋਈ ਇੰਟਰਫੇਸ ਹੁੰਦੀ ਹੈ ਜਿਵੇਂ ਕਿ ਹੇਠਾਂ ਦਿੱਤਾ ਗਿਆ:

- ਕਮਾਂਡ ਅਧਾਰਿਤ। ਕਨਸੋਲ ਐਪ ਉਹ ਹੁੰਦੀਆਂ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਇਹ ਕੰਮ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਣ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪ ਵਿਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUIs) ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਇਣਪੁੱਟ ਕਰਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਆਦਿ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪ ਸੀਮਿਤ ਹੁੰਦੀਆਂ ਹਨ

ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਨਾਲ ਤੁਲਨਾ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ:

- **ਇਹ ਸੀਮਿਤ ਹੈ**। ਤੁਸੀਂ ਹਰੇਕ ਕਮਾਂਡ ਨਹੀਂ ਲਿਖ ਸਕਦੇ, ਸਿਰਫ ਉਹ ਕਮਾਂਡਾਂ ਜੋ ਐਪ ਸਹਾਰਦਾ ਹੋਵੇ।
- **ਭਾਸ਼ਾ ਵਿਸ਼ੇਸ਼**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਰਦੇ ਹਨ, ਪਰ ਮੂਲ ਰੂਪ ਵਿੱਚ ਐਪ ਕਿਸੇ ਇੱਕ ਵਿਸ਼ੇਸ਼ ਭਾਸ਼ਾ ਲਈ ਬਣਾਇਆ ਜਾਂਦਾ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਲਈ ਸਹਾਰਾ ਜੋੜ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਦੇ ਫਾਇਦੇ

ਤਾਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੁੰਦਾ ਹੈ?

ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ, ਤੁਹਾਡੇ ਕੋਲ ਵੱਧ ਲਚੀਲਾਪਨ ਹੁੰਦਾ ਹੈ, ਤੁਸੀਂ ਕਿਸੇ ਨਿਰਧਾਰਿਤ ਕਮਾਂਡ ਜਾਂ ਵਿਸ਼ੇਸ਼ ਇਣਪੁੱਟ ਭਾਸ਼ਾ ਨਾਲ ਸੀਮਿਤ ਨਹੀਂ ਹੁੰਦੇ। ਇਸ ਦੇ ਬਦਲੇ ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਐਪ ਨਾਲ ਗੱਲਬਾਤ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਪਹਿਲਾਂ ਹੀ ਉਸ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਗੱਲਬਾਤ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੇ ਜਾਣਕਾਰੀਆਂ ਦੇ ਸੰਗ੍ਰਹਿ ਤੇ ਪ੍ਰਸ਼িক্ষਿਤ ਹੈ, ਜਦਕਿ ਇੱਕ ਪਰੰਪਰਾਗਤ ਐਪ ਵਿੱਚ ਡਾਟਾਬੇਸ ਵਿੱਚ ਕੀ ਹੈ ਉਸ 'ਤੇ ਸੀਮਤ ਹੋ ਸਕਦਾ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂ ਸਕਦਾ ਹਾਂ?

ਤੁਹਾਨੂੰ ਕਈ ਚੀਜ਼ਾਂ ਬਣਾਉਣ ਲਈ ਮਿਲ ਸਕਦੀਆਂ ਹਨ। ਉਦਾਹਰਣ ਵਜੋਂ:

- **ਇੱਕ ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਵਿਸ਼ਿਆਂ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਉੱਤਰ ਦਿੰਦਾ ਹੈ, ਜਿਵੇਂ ਤੁਹਾਡੀ ਕੰਪਨੀ ਅਤੇ ਇਸਦੇ ਉਤਪਾਦ, ਇੱਕ ਵਧੀਆ ਮੇਲ ਹੋ ਸਕਦਾ ਹੈ।
- **ਸਹਾਇਕ**। LLMs ਟੈਕਸਟ ਸੰਖੇਪ ਕਰਨ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀ ਹਾਸਲ ਕਰਨ, ਰੇਜ਼ਿਊਮੇ ਅਤੇ ਹੋਰ ਵਰਗਾ ਟੈਕਸਟ ਬਣਾਉਣ ਵਿੱਚ ਵਧੀਆ ਹੁੰਦੇ ਹਨ।
- **ਕੋਡ ਸਹਾਇਕ**। ਭਾਸ਼ਾ ਮਾਡਲ ਦੇ ਅਧਾਰ 'ਤੇ, ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਸਹਾਇਕ ਬਣਾਉਣ ਵਾਸਤੇ ਜੋ ਤੁਹਾਡੇ ਲਈ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਣ ਵਜੋਂ, ਤੁਸੀਂ GitHub Copilot ਜਾਂ ChatGPT ਵਰਗੇ ਪ੍ਰੋਡਕਟ ਦੀ ਵਰਤੋਂ ਕੋਡ ਲਿਖਣ ਲਈ ਕਰ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਖੈਰ, ਤੁਹਾਨੂੰ ਕਿਸੇ LLM ਨਾਲ ਇਕੱਠੇ ਕੰਮ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪੈਂਦਾ ਹੈ ਜੋ ਆਮ ਤੌਰ 'ਤੇ ਹੇਠਾਂ ਦਿੱਤੇ ਗਏ ਦੋ ਤਰੀਕਿਆਂ ਵਿੱਚੋਂ ਇਕ ਹੁੰਦਾ ਹੈ:

- API ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣੇ ਪ੍ਰਾਪੰਪਟ ਨਾਲ ਵੈੱਬ ਰਿਕਵੇਸਟ ਤਿਆਰ ਕਰਦੇ ਹੋ ਅਤੇ ਬਣਿਆ ਹੋਇਆ ਟੈਕਸਟ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਾਂ ਨੂੰ ਇਕੱਠਾ ਕਰਦੀਆਂ ਹਨ ਅਤੇ ਇਹਨਾਂ ਨੂੰ ਵਰਤਣਾ ਆਸਾਨ ਬਣਾ ਦਿੰਦੀਆਂ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀ/SDKs

LLMs ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਪ੍ਰਸਿੱਧ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰਾਪੰਪਟ ਭੇਜਣ ਲਈ ਆਸਾਨੀ ਹੈ।

ਫਿਰ ਉਚ-ਪੱਧਰ 'ਤੇ ਕੰਮ ਕਰਨ ਵਾਲੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **Langchain**। Langchain ਮਸ਼ਹੂਰ ਹੈ ਅਤੇ Python ਨੂੰ ਸਹਾਰਦਾ ਹੈ।
- **Semantic Kernel**। Semantic Kernel ਮਾਈਕ੍ਰੋਸਾਫਟ ਦੀ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python, ਅਤੇ Java ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਰਦੀ ਹੈ।

## ChatCompletions ਨਾਲ ਟੈਕਸਟ ਬਣਾਓ

ਟੈਕਸਟ ਬਣਾਉਣ ਦਾ ਤਰੀਕਾ `ChatCompletionsClient` ਕਲਾਸ ਦੀ ਵਰਤੋਂ ਕਰਨਾ ਹੈ।
`samples/python/azure_ai_inference/basic.py` ਵਿੱਚ, ਕੋਡ ਦੇ ਜਵਾਬ ਵਿੱਚ, ਉਪਭੋਗਤਾ ਭੂਮਿਕਾ ਵਿੱਚ “content” ਪੈਰਾਮੀਟਰ ਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਇਸ ਤਰੀਕੇ ਨਾਲ ਅੱਪਡੇਟ ਕਰੋ:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ਵਿਕਲਪਿਕ ਪੈਰਾਮੀਟਰ
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

ਅੱਪਡੇਟ ਕੀਤੀ ਫਾਇਲ ਚਲਾਓ ਅਤੇ ਨਤੀਜਾ ਵੇਖੋ

1. ਇੱਕ `prompt` ਵੈਰੀਏਬਲ ਬਣਾਓ ਅਤੇ ਨਮੂਨਾ ਕੋਡ ਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਅਨੁਸਾਰ ਬਦਲੋ:
