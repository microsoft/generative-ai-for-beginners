# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਪਾਠਕ੍ਰਮ ਰਾਹੀਂ ਦੇਖਿਆ ਹੈ ਕਿ ਕੁਝ ਮੁੱਖ ਧਾਰਾਵਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰੌੰਪਟ ਅਤੇ ਇੱਥੋਂ ਤੱਕ ਕਿ "ਪ੍ਰੌੰਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਨਾਂ ਦੀ ਪੂਰੀ ਵਿਧਾ ਵੀ ਹੈ। ਕਈ ਟੂਲ ਜਿਵੇਂ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ ਤੁਹਾਨੂੰ ਪ੍ਰੌੰਪਟ ਵਰਤ ਕੇ ਕੁਝ ਹਾਸਲ ਕਰਨ ਵਿੱਚ ਮਦਦ ਕਰਦੇ ਹਨ।

ਜੇ ਤੁਸੀਂ ਆਪਣੀ ਐਪ ਵਿੱਚ ਇਹ ਤਜਰਬਾ ਸ਼ਾਮਲ ਕਰਨਾ ਚਾਹੁੰਦੇ ਹੋ, ਤਾਂ ਤੁਹਾਨੂੰ ਪ੍ਰੌੰਪਟ, ਕੰਪਲੀਸ਼ਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਅਤੇ ਵਰਤਣ ਲਈ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣਨੀਆਂ ਆਉਣੀਆਂ ਚਾਹੀਦੀਆਂ ਹਨ। ਇਹੀ ਗੱਲ ਤੁਸੀਂ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖੋਗੇ।

## ਜਾਣ ਪਛਾਣ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਿੱਖੋਗੇ।
- ਪ੍ਰੌੰਪਟ, ਟੈਮਪਰੇਚਰ ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ 'ਤੇ, ਤੁਸੀਂ ਸਮਝ ਸਕੋਗੇ:

- ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ।

- ਆਪਣੀ ਐਪ ਵਿੱਚ ਟੋਕਨ ਦੀ ਗਿਣਤੀ ਵਧਾਉਣ-ਘਟਾਉਣ ਅਤੇ ਟੈਮਪਰੇਚਰ ਬਦਲਣ ਦੀ ਸੰਰਚਨਾ ਕਰ ਸਕਦੇ ਹੋ, ਤਾਂ ਜੋ ਨਤੀਜਾ ਵੱਖ-ਵੱਖ ਆ ਸਕੇ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਕੋਈ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਉਸਦਾ ਕੁਝ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ, ਜਿਵੇਂ:

- ਕਮਾਂਡ-ਅਧਾਰਤ। ਕੰਸੋਲ ਐਪ ਆਮ ਤੌਰ 'ਤੇ ਉਹ ਐਪ ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਉਹ ਕੰਮ ਕਰ ਦਿੰਦੇ ਹਨ। ਉਦਾਹਰਨ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪਾਂ ਵਿੱਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUI) ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਲਿਖਦੇ ਹੋ, ਚੋਣ ਕਰਦੇ ਹੋ ਆਦਿ।

### ਕੰਸੋਲ ਅਤੇ UI ਐਪਾਂ ਦੀਆਂ ਹੱਦਾਂ

ਇਸਨੂੰ ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਨਾਲ ਤੁਲਨਾ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ:

- **ਇਹ ਸੀਮਤ ਹੈ**। ਤੁਸੀਂ ਹਰ ਕਿਸੇ ਵੀ ਕਮਾਂਡ ਨਹੀਂ ਲਿਖ ਸਕਦੇ, ਸਿਰਫ ਉਹੀ ਜੋ ਐਪ ਸਹਾਇਕ ਕਰਦੀ ਹੈ।
- **ਭਾਸ਼ਾ-ਨਿਰਧਾਰਤ**। ਕੁਝ ਐਪਾਂ ਕਈ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਕ ਕਰਦੀਆਂ ਹਨ, ਪਰ ਆਮ ਤੌਰ 'ਤੇ ਐਪ ਇੱਕ ਖਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣੀ ਹੁੰਦੀ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਜੋੜ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਾਂ ਦੇ ਫਾਇਦੇ

ਫਿਰ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰੀ ਹੈ?

ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਹੋਰ ਲਚਕ ਹੁੰਦੀ ਹੈ, ਤੁਸੀਂ ਸਿਰਫ਼ ਕੁਝ ਕਮਾਂਡਾਂ ਜਾਂ ਖਾਸ ਭਾਸ਼ਾ ਤੱਕ ਸੀਮਤ ਨਹੀਂ ਹੋ। ਇਸ ਦੀ ਥਾਂ, ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਰਾਹੀਂ ਐਪ ਨਾਲ ਗੱਲ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਇੱਕ ਐਸੇ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਗੱਲ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੀ ਜਾਣਕਾਰੀ 'ਤੇ ਟਰੇਨ ਹੋਇਆ ਹੈ, ਜਦਕਿ ਰਵਾਇਤੀ ਐਪ ਸਿਰਫ਼ ਡਾਟਾਬੇਸ ਤੱਕ ਸੀਮਤ ਹੋ ਸਕਦੀ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂ ਸਕਦਾ ਹਾਂ?

ਤੁਸੀਂ ਕਈ ਚੀਜ਼ਾਂ ਬਣਾਉਣੀਆਂ। ਉਦਾਹਰਨ ਵਜੋਂ:

- **ਚੈਟਬੋਟ**। ਚੈਟਬੋਟ ਜੋ ਕਿਸੇ ਵਿਸ਼ੇ, ਜਿਵੇਂ ਤੁਹਾਡੀ ਕੰਪਨੀ ਜਾਂ ਉਤਪਾਦਾਂ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇਵੇ, ਵਧੀਆ ਚੋਣ ਹੋ ਸਕਦੀ ਹੈ।
- **ਹੈਲਪਰ**। LLMs ਟੈਕਸਟ ਸੰਖੇਪ ਕਰਨ, ਟੈਕਸਟ ਤੋਂ ਝਲਕੀਆਂ ਲੈਣ, ਰਿਜ਼ਿਊਮੇ ਵਰਗਾ ਟੈਕਸਟ ਬਣਾਉਣ ਆਦਿ ਵਿੱਚ ਵਧੀਆ ਹਨ।
- **ਕੋਡ ਅਸਿਸਟੈਂਟ**। ਤੁਸੀਂ ਜਿਸ ਭਾਸ਼ਾ ਮਾਡਲ ਦੀ ਵਰਤੋਂ ਕਰਦੇ ਹੋ, ਉਸ ਅਨੁਸਾਰ ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਅਸਿਸਟੈਂਟ ਵੀ ਬਣਾਉਣੀ ਜੋ ਤੁਹਾਨੂੰ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰੇ। ਉਦਾਹਰਨ ਵਜੋਂ, ਤੁਸੀਂ GitHub Copilot ਜਾਂ ChatGPT ਵਰਗੇ ਉਤਪਾਦ ਵਰਤ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰਾਂ?

ਤੁਹਾਨੂੰ LLM ਨਾਲ ਇੰਟੀਗਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪਵੇਗਾ, ਆਮ ਤੌਰ 'ਤੇ ਇਹ ਦੋ ਤਰੀਕਿਆਂ 'ਚੋਂ ਇੱਕ ਹੁੰਦਾ ਹੈ:

- API ਵਰਤੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣਾ ਪ੍ਰੌੰਪਟ ਲੈ ਕੇ ਵੈੱਬ ਰਿਕਵੈਸਟ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਵਾਪਸ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਮਿਲਦਾ ਹੈ।
- ਲਾਇਬ੍ਰੇਰੀ ਵਰਤੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਾਂ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀਆਂ ਹਨ।

## ਕਸਰਤ - ਇੱਕ ਰੈਸੀਪੀ ਜਨਰੇਟਰ ਬਣਾਓ

ਹੁਣ ਜਦੋਂ ਅਸੀਂ ਇੱਕ ਸਥਿਤੀ 'ਤੇ ਖੇਡ ਚੁੱਕੇ ਹਾਂ, ਆਓ ਉਸ ਦਿਖਾਏ ਗਏ ਸਥਿਤੀ ਨੂੰ ਮਿਲਾਉਣ ਲਈ ਕੋਡ ਲਿਖੀਏ। ਇਹ ਕਰਨ ਲਈ, ਹੇਠਾਂ ਦਿੱਤੇ ਕਦਮਾਂ ਦੀ ਪਾਲਣਾ ਕਰੋ:

1. ਮੌਜੂਦਾ ਫਾਈਲ ਨੂੰ ਸ਼ੁਰੂਆਤ ਵਜੋਂ ਵਰਤੋ
1. ਇੱਕ `prompt` ਵੈਰੀਏਬਲ ਬਣਾਓ ਅਤੇ ਨਮੂਨਾ ਕੋਡ ਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਤਰੀਕੇ ਨਾਲ ਬਦਲੋ:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
