# ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਕੋਰਸ ਵਿੱਚ ਦੇਖਿਆ ਹੈ ਕਿ ਇਸ ਵਿੱਚ ਬੁਨਿਆਦੀ ਧਾਰਨਾਵਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰਾਂਪਟਸ ਅਤੇ ਇੱਕ ਪੂਰਾ ਵਿਸ਼ਾ ਜਿਸ ਨੂੰ "ਪ੍ਰਾਂਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਕਹਿੰਦੇ ਹਨ। ਕਈ ਤਰ੍ਹਾਂ ਦੇ ਟੂਲ ਜਿਵੇਂ ਕਿ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ, ਤੁਹਾਨੂੰ ਪ੍ਰਾਂਪਟਸ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਕੁਝ ਪ੍ਰਾਪਤ ਕਰਨ ਵਿੱਚ ਸਹਾਇਤਾ ਦਿੰਦੇ ਹਨ।

ਤੁਹਾਨੂੰ ਐਪ ਵਿੱਚ ਐਸਾ ਅਨੁਭਵ ਸ਼ਾਮਲ ਕਰਨ ਲਈ, ਤੁਹਾਨੂੰ ਪ੍ਰਾਂਪਟਸ, ਪੂਰਨਾਂ ਅਤੇ ਕੰਮ ਕਰਨ ਲਈ ਇੱਕ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਚੋਣ ਵਰਗੀਆਂ ਧਾਰਨਾਵਾਂ ਨੂੰ ਸਮਝਣਾ ਲਾਜ਼ਮੀ ਹੈ। ਇੱਥੇ ਤੁਹਾਡੇ ਲਈ ਇਹੀ ਸਿਖਾਉਣਾ ਹੈ।

## ਪਰਿਚਯ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸ ਦੀਆਂ ਬੁਨਿਆਦੀ ਧਾਰਨਾਵਾਂ ਬਾਰੇ ਸਿੱਖੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਗੇ।
- ਸਮਝੋਗੇ ਕਿ ਕਿਵੇਂ ਪ੍ਰਾਂਪਟ, ਤਾਪਮਾਨ, ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਧਾਰਨਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਤਿਆਰ ਕਰੀਦਾ ਹੈ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਵਿੱਚ, ਤੁਸੀਂ ਯੋਗ ਹੋਵੋਗੇ:

- ਇਹ ਸਮਝਾਉਣਾ ਕਿ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣਾ।
- ਐਪ ਨੂੰ ਇਸ ਤਰ੍ਹਾਂ ਕੰਫਿਗਰ ਕਰਨਾ ਕਿ ਇਸ ਵਿੱਚ ਵੱਧ ਜਾਂ ਘੱਟ ਟੋਕਨ ਵਰਤਣ ਦੇ ਨਾਲ ਨਾਲ ਤਾਪਮਾਨ ਨੂੰ ਵੀ ਬਦਲਿਆ ਜਾ ਸਕੇ, ਤਾਂ ਜੋ ਵੱਖ-ਵੱਖ ਨਤੀਜੇ ਆ ਸਕਣ।

## ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ, ਜਦੋਂ ਤੁਸੀਂ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਇਹ ਵਿੱਚ ਕਿਸੇ ਕਿਸਮ ਦਾ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ ਜਿਵੇਂ ਕਿ:

- ਕਮਾਂਡ-ਆਧਾਰਿਤ। ਕੰਸੋਲ ਐਪ ਆਮ ਤੌਰ 'ਤੇ ਐਸੀਆਂ ਐਪ ਹੁੰਦੀਆਂ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਿਸੇ ਕਮਾਂਡ ਨੂੰ ਟਾਈਪ ਕਰਦੇ ਹੋ ਅਤੇ ਇਹ ਕੰਮ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਲਈ, `git` ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਹੈ।
- ਉਪਭੋਗਤਾ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUIs) ਰੱਖਦੀਆਂ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨਾਂ ਨੂੰ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਡਾਲਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਅਤੇ ਹੋਰ।

### ਕੰਸੋਲ ਅਤੇ UI ਐਪ ਸੀਮਿਤ ਹਨ

ਇਸ ਦੀ ਤੁਲਨਾ ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਨਾਲ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ:

- **ਇਹ ਸੀਮਿਤ ਹੈ**। ਤੁਸੀਂ ਕਿਰਪਾ ਕਰਕੇ ਕੋਈ ਵੀ ਕਮਾਂਡ ਟਾਈਪ ਨਹੀਂ ਕਰ ਸਕਦੇ, ਸਿਰਫ਼ ਉਹ ਜੋ ਐਪ ਸਹਾਇਤਾ ਕਰਦੀ ਹੈ।
- **ਭਾਸ਼ਾ ਖਾਸ**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦੇ ਹਨ ਪਰ ਡਿਫਾਲਟ ਰੂਪ ਵਿੱਚ ਇਹ ਕਿਸੇ ਖਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣਾਈਆਂ ਜਾਂਦੀਆਂ ਹਨ, ਹਾਲਾਂਕਿ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਤਾ ਵਿੱਚ ਸ਼ਾਮਲ ਕਰ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਦੇ ਫਾਇਦੇ

ਤਾਂ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੁੰਦੀ ਹੈ?

ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਵੱਧ ਲਚਕੀਲਾਪਣ ਹੁੰਦੀ ਹੈ, ਤੁਸੀਂ ਕਿਸੇ ਪੱਕੀ ਕਮਾਂਡ ਜਾਂ ਖਾਸ ਇਨਪੁਟ ਭਾਸ਼ਾ ਤੱਕ ਸੀਮਿਤ ਨਹੀਂ ਹੁੰਦੇ। ਇਸਦੀ ਬਜਾਏ, ਤੁਸੀਂ ਸਧਾਰਣ ਭਾਸ਼ਾ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਐਪ ਨਾਲ ਗੱਲ-ਬਾਤ ਕਰ ਸਕਦੇ ਹੋ। ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਪਹਿਲਾਂ ਹੀ ਇੱਕ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਸੰਪਰਕ ਵਿੱਚ ਹੋ ਜੋ ਵੱਡੀ ਮਾਤਰਾ ਵਿੱਚ ਜਾਣਕਾਰੀ 'ਤੇ ਤਿਆਰ ਕੀਤਾ ਗਿਆ ਹੈ, ਜਦਕਿ ਇੱਕ ਪਰੰਪਰਾਗਤ ਐਪ ਡਾਟਾਬੇਸ ਵਿੱਚ ਮੌਜੂਦ ਸਮੱਗਰੀ 'ਤੇ ਸੀਮਿਤ ਹੋ ਸਕਦੀ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜੈਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਇ ਸਕਦਾ ਹਾਂ?

ਤੁਸੀਂ ਕਈ ਚੀਜ਼ਾਂ ਬਣਾ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਲਈ:

- **ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਤੁਹਾਡੇ ਕੰਪਨੀ ਅਤੇ ਇਸਦੇ ਉਤਪਾਦਾਂ ਬਾਰੇ ਪ੍ਰਸ਼ਨਾਂ ਦੇ ਜਵਾਬ ਦੇ ਸਕਦਾ ਹੈ।
- **ਸਹਾਇਕ**। LLMs ਟੈਕਸਟ ਨੂੰ ਸੰਖੇਪ ਕਰਨਾ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀ ਪ੍ਰਾਪਤ ਕਰਨਾ, ਰੈਜ਼ੂਮੇ ਆਦਿ ਜੈਸੀ ਚੀਜ਼ਾਂ ਬਣਾਉਣਾ ਵਿੱਚ ਮਹਾਨ ਹਨ।
- **ਕੋਡ ਸਹਾਇਕ**। ਤੁਸੀਂ ਜੋ ਭਾਸ਼ਾ ਮਾਡਲ ਵਰਤਦੇ ਹੋ ਉਸਦੇ ਅਨੁਸਾਰ ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਸਹਾਇਕ ਬਣਾ ਸਕਦੇ ਹੋ ਜੋ ਤੁਹਾਨੂੰ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਲਈ, ਤੁਸੀਂ GitHub Copilot ਜਿਵੇਂ ਉਤਪਾਦਾਂ ਨਾਲ ਨਾਲ ChatGPT ਦੀ ਵੀ ਵਰਤੋਂ ਕਰਕੇ ਕੋਡ ਲਿਖਣ ਵਿਚ ਸਹਾਇਤਾ ਲੈ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਬਿਲਕੁਲ, ਤੁਹਾਨੂੰ ਇੱਕ LLM ਨਾਲ ਇੰਟੀਗਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪੈਂਦਾ ਹੈ ਜੋ ਆਮ ਤੌਰ 'ਤੇ ਹੇਠਾਂ ਦਿੱਤੇ ਦੋ ਤਰੀਕਿਆਂ ਵਿਚੋਂ ਹੋ ਸਕਦਾ ਹੈ:

- API ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣੇ ਪ੍ਰਾਂਪਟ ਨਾਲ ਵੈੱਬ ਬੇਨਤੀਆਂ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਜੈਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀਜ਼ API ਕਾਲਜ਼ ਨੂੰ ਲਪੇਟ ਕੇ ਉਨ੍ਹਾਂ ਨੂੰ ਵਰਤਣਾ ਸੌਖਾ ਬਣਾ ਦਿੰਦੇ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀਜ਼/SDKs

LLMs ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਪ੍ਰਸਿੱਧ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ ਕਿ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰਾਂਪਟ ਭੇਜਣ ਵਿੱਚ ਅਸਾਨੀ ਪੈਦਾ ਕਰਦੀ ਹੈ।

ਫਿਰ ਕੁਝ ਉੱਚ ਪੱਧਰ ਦੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਵੀ ਹਨ:

- **Langchain**। Langchain ਪ੍ਰਸਿੱਧ ਹੈ ਅਤੇ ਪਾਈਥਨ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦਾ ਹੈ।
- **Semantic Kernel**। Semantic Kernel ਮਾਈਕ੍ਰੋਸਾਫਟ ਦੀ ਇੱਕ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, ਪਾਈਥਨ, ਅਤੇ ਜਾਵਾ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦੀ ਹੈ।

## openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਪਹਿਲੀ ਐਪ

ਆਓ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਆਪਣੀ ਪਹਿਲੀ ਐਪ ਕਿਵੇਂ ਬਣਾਉਣੀ ਹੈ, ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਦੀ ਲੋੜ ਹੈ, ਕਿੰਨਾ ਚਾਹੀਦਾ ਹੈ ਆਦਿ।

### openai ਇੰਸਟਾਲ ਕਰੋ

  > [!NOTE] ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Devcontainer ਵਿੱਚ ਚਲਾ ਰਹੇ ਹੋ ਤਾਂ ਇਹ ਕਦਮ ਜ਼ਰੂਰੀ ਨਹੀਂ।


OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਇੰਟਰਐਕਟ ਕਰਨ ਲਈ ਕਈ ਲਾਇਬ੍ਰੇਰੀਆਂ ਮੌਜੂਦ ਹਨ। C#, Python, JavaScript, Java ਵਰਗੀਆਂ ਕਈ ਪ੍ਰੋਗ੍ਰਾਮਿੰਗ ਭਾਸ਼ਾਵਾਂ ਦੀ ਵਰਤੋਂ ਵੀ ਕੀਤੀ ਜਾ ਸਕਦੀ ਹੈ।  
ਅਸੀਂ `openai` ਪਾਈਥਨ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਦੀ ਚੋਣ ਕੀਤੀ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `pip` ਨਾਲ ਇਸ ਨੂੰ ਇੰਸਟਾਲ ਕਰਾਂਗੇ।

```bash
pip install openai
```

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Devcontainer ਵਿੱਚ ਨਹੀਂ ਚਲਾ ਰਹੇ, ਤਾਂ ਤੁਹਾਨੂੰ ਆਪਣੇ ਕੰਪਿਊਟਰ 'ਤੇ [Python](https://www.python.org/) ਵੀ ਇੰਸਟਾਲ ਕਰਨੀ ਹੋਵੇਗੀ।

### ਇੱਕ ਰਿਸੋਰਸ ਬਣਾਓ ਅਤੇ ਆਪਣੀ API ਕੀ ਲੱਭੋ

ਜੇ ਤੁਸੀਂ ਪਹਿਲਾਂ ਹੀ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ ਇਨ੍ਹਾਂ ਕਦਮਾਂ ਨੂੰ ਪੂਰਾ ਕਰੋ:

- OpenAI <https://platform.openai.com/signup> 'ਤੇ ਇੱਕ ਅਕਾਊਂਟ ਬਣਾਓ।
- ਹੁਣ, ਆਪਣੀਆਂ API ਕੁੰਜੀਆਂ ਪ੍ਰਾਪਤ ਕਰੋ <https://platform.openai.com/api-keys>।

>[!NOTE]
> ਆਪਣੀ API ਕੀ ਨੂੰ ਆਪਣੇ ਕੋਡ ਤੋਂ ਅਲੱਗ ਰੱਖਣਾ ਵਧੀਆ ਹੈ। ਤੁਸੀਂ ਇਹ ਮਾਹੌਲ ਵੈਰੀਏਬਲਜ਼ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਕਰ ਸਕਦੇ ਹੋ।
> - ਆਪਣੀ .env ਫਾਈਲ ਵਿੱਚ ਮਾਹੌਲ ਵੈਰੀਏਬਲ `OPENAI_KEY` ਨੂੰ ਆਪਣੇ API ਕੁੰਜੀ ਨਾਲ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਇਸ ਕੋਰਸ ਦੀਆਂ ਪਹਿਲੀਆਂ ਕਸਰਤਾਂ ਮੁਕੰਮਲ ਕਰ ਲਈਆਂ ਹਨ, ਤਾਂ ਤੁਸੀਂ ਤਿਆਰ ਹੋ।
> - ਯਾਦ ਰਹੇ ਕਿ API ਕੀ ਸਿਰਫ ਇਕ ਵਾਰ ਮਿਲਦੀ ਹੈ। ਇਸ ਲਈ ਜ਼ਰੂਰੀ ਹੈ ਕਿ ਤੁਸੀਂ ਇਸਦੀ ਸਹੀ ਨਕਲ ਕਰੋ। ਜੇ ਇਹ ਸਹੀ ਤਰ੍ਹਾਂ ਕੰਮ ਨਾ ਕਰੇ, ਤਾਂ ਕੀ ਨੂੰ ਮਿਟਾ ਕੇ ਨਵੀਂ ਬਣਾਉ।


### OpenAI ਸੰਰਚਨਾ ਸੈਟਅਪ ਕਰੋ

ਜੇ ਤੁਸੀਂ OpenAI ਵਰਤ ਰਹੇ ਹੋ, ਤਾਂ ਇਸ ਤਰ੍ਹਾਂ ਸੰਰਚਨਾ ਸੈਟ ਕਰੋ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

## ਵਰਕਆਉਟ - ਇਕ ਰੈਸਪੀ ਜਨਰੇਟਰ ਬਣਾਓ

1. ਮੌਜੂਦਾ *app.py* ਫਾਇਲ ਨੂੰ ਸ਼ੁਰੂਆਤੀ ਬਿੰਦੂ ਵਜੋਂ ਵਰਤੋ

ਵਧੀਆ, ਆਓ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਗੱਲਾਂ ਕਿਵੇਂ ਸੁਧਾਰ ਸਕਦੇ ਹਾਂ। ਗੱਲਾਂ ਨੂੰ ਸੁਧਾਰਨ ਲਈ, ਅਸੀਂ ਯਕੀਨ ਕਰਨਾ ਚਾਹੁੰਦੇ ਹਾਂ ਕਿ ਕੋਡ ਲਚਕੀਲਾ ਹੋਵੇ, ਤਾਂ ਜੋ ਸਮੱਗਰੀ ਅਤੇ ਨੁਸਖਿਆਂ ਦੀ ਗਿਣਤੀ ਬਿਹਤਰ ਅਤੇ ਬਦਲੀ ਜਾ ਸਕੇ।

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- ਰੈਸਿਪੀ ਜਨਰੇਟਰ ਐਪ ਨੂੰ ਹੋਰ ਸੁਧਾਰਨ ਲਈ ਟਵੀਕ ਕਰੋ। ਤਾਪਮਾਨ ਮੁੱਲਾਂ ਨਾਲ ਖੇਡੋ, ਅਤੇ ਪ੍ਰਾਂਪਟਾਂ ਨੂੰ ਦੇਖੋ ਕਿ ਤੁਸੀਂ ਕੀ ਕੱਢ ਸਕਦੇ ਹੋ।
- ਇੱਕ "ਸਟਡੀ ਬੱਡੀ" ਬਣਾਓ। ਇਸ ਐਪ ਨੂੰ ਕਿਸੇ ਵਿਸ਼ੇ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇਣ ਯੋਗ ਹੋਣਾ ਚਾਹੀਦਾ ਹੈ, ਉਦਾਹਰਨ ਵਜੋਂ ਪਾਇਥਨ, ਤੁਸੀਂ ਪ੍ਰਾਂਪਟਾਂ ਜਿਵੇਂ "ਪਾਇਥਨ ਵਿੱਚ ਕਿਸੇ ਵਿਸ਼ੇ ਦਾ ਕੀ ਹੈ?", ਜਾਂ ਤੁਸੀਂ ਇੱਕ ਐਸਾ ਪ੍ਰਾਂਪਟ ਰੱਖ ਸਕਦੇ ਹੋ ਜੋ ਕਿਸੇ ਵਿਸ਼ੇ ਲਈ ਕੋਡ ਵੇਖਾਏ ਆਦਿ।
- ਹਿਸਟਰੀ ਬੌਟ, ਇਤਿਹਾਸ ਨੂੰ ਜੀਵੰਤ ਬਣਾਓ, ਬੌਟ ਨੂੰ ਕਿਸੇ ਇਤਿਹਾਸਕ ਪাত্র ਦਾ ਨਿਰਦੇਸ਼ ਦਿੱਤੋ ਅਤੇ ਇਸਦੇ ਜੀਵਨ ਅਤੇ ਸਮਿਆਂ ਬਾਰੇ ਸਵਾਲ ਪੁੱਛੋ।
