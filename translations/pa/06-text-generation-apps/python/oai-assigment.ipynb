# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਪਾਠਕ੍ਰਮ ਵਿੱਚ ਦੇਖਿਆ ਕਿ ਕੁਝ ਮੁੱਖ ਧਾਰਾਵਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰੋੰਪਟ ਅਤੇ ਇੱਕ ਪੂਰੀ ਵਿਧਾ "ਪ੍ਰੋੰਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਵੀ ਹੈ। ਕਈ ਟੂਲ ਜਿਵੇਂ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ ਤੁਹਾਨੂੰ ਪ੍ਰੋੰਪਟ ਵਰਤ ਕੇ ਕੰਮ ਕਰਨ ਵਿੱਚ ਮਦਦ ਕਰਦੇ ਹਨ।

ਜੇ ਤੁਸੀਂ ਆਪਣੇ ਐਪ ਵਿੱਚ ਇਹ ਤਜਰਬਾ ਸ਼ਾਮਲ ਕਰਨਾ ਚਾਹੁੰਦੇ ਹੋ, ਤਾਂ ਤੁਹਾਨੂੰ ਪ੍ਰੋੰਪਟ, ਕੰਪਲੀਸ਼ਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਅਤੇ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣਨੀਆਂ ਪੈਂਦੀਆਂ ਹਨ। ਇਹੀ ਗੱਲ ਤੁਸੀਂ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖੋਗੇ।

## ਪਰਚਿਆ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸ ਦੀਆਂ ਮੁੱਖ ਧਾਰਾਵਾਂ ਬਾਰੇ ਜਾਣੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਿੱਖੋਗੇ।
- ਪ੍ਰੋੰਪਟ, ਟੈਮਪਰੇਚਰ, ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਿੱਖੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ 'ਤੇ, ਤੁਸੀਂ:

- ਸਮਝਾ ਸਕੋਗੇ ਕਿ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਿੱਖੋਗੇ।
- ਆਪਣੇ ਐਪ ਵਿੱਚ ਟੋਕਨ ਦੀ ਗਿਣਤੀ ਵਧਾ-ਘਟਾ ਸਕੋਗੇ ਅਤੇ ਟੈਮਪਰੇਚਰ ਵੀ ਬਦਲ ਸਕੋਗੇ, ਤਾਂ ਜੋ ਨਤੀਜਾ ਵੱਖ-ਵੱਖ ਆ ਸਕੇ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਐਪ ਬਣਾਉਂਦੇ ਹੋ, ਤਾਂ ਇਸ ਦੀ ਕੋਈ ਨਾ ਕੋਈ ਇੰਟਰਫੇਸ ਹੁੰਦੀ ਹੈ, ਜਿਵੇਂ:

- ਕਮਾਂਡ-ਅਧਾਰਤ। ਕਨਸੋਲ ਐਪ ਆਮ ਤੌਰ 'ਤੇ ਉਹ ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਇਹ ਕੰਮ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪ ਵਿੱਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUI) ਹੁੰਦੇ ਹਨ, ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਲਿਖਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਆਦਿ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪ ਦੀਆਂ ਸੀਮਾਵਾਂ

ਇਸਨੂੰ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਨਾਲ ਤੁਲਨਾ ਕਰੋ, ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ:

- **ਇਹ ਸੀਮਤ ਹੈ**। ਤੁਸੀਂ ਹਰ ਕਿਸਮ ਦੀ ਕਮਾਂਡ ਨਹੀਂ ਲਿਖ ਸਕਦੇ, ਸਿਰਫ ਉਹੀ ਜੋ ਐਪ ਸਹਾਇਕ ਕਰਦੀ ਹੈ।
- **ਭਾਸ਼ਾ-ਵਿਸ਼ੇਸ਼**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਸਹਾਇਕ ਕਰਦੇ ਹਨ, ਪਰ ਆਮ ਤੌਰ 'ਤੇ ਐਪ ਇੱਕ ਖਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣੀ ਹੁੰਦੀ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਸ਼ਾਮਲ ਕਰ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਦੇ ਫਾਇਦੇ

ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੈ?

ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਹੋਰ ਲਚੀਲਤਾ ਹੁੰਦੀ ਹੈ, ਤੁਸੀਂ ਕਿਸੇ ਖਾਸ ਕਮਾਂਡ ਜਾਂ ਇੰਪੁੱਟ ਭਾਸ਼ਾ ਤੱਕ ਸੀਮਤ ਨਹੀਂ ਹੁੰਦੇ। ਇਸ ਦੀ ਬਜਾਏ, ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਵਿੱਚ ਐਪ ਨਾਲ ਗੱਲ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਇੱਕ ਐਸੇ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਗੱਲ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੀ ਜਾਣਕਾਰੀ 'ਤੇ ਟ੍ਰੇਨ ਹੋਇਆ ਹੈ, ਜਦਕਿ ਰਵਾਇਤੀ ਐਪ ਸਿਰਫ ਡਾਟਾਬੇਸ ਤੱਕ ਸੀਮਤ ਹੋ ਸਕਦੀ ਹੈ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਇਆ ਜਾ ਸਕਦਾ ਹੈ?

ਕਈ ਚੀਜ਼ਾਂ ਬਣਾਈਆਂ ਜਾ ਸਕਦੀਆਂ ਹਨ। ਉਦਾਹਰਨ ਵਜੋਂ:

- **ਚੈਟਬੋਟ**। ਚੈਟਬੋਟ ਜੋ ਕਿਸੇ ਵਿਸ਼ੇ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇਵੇ, ਜਿਵੇਂ ਤੁਹਾਡੀ ਕੰਪਨੀ ਅਤੇ ਇਸਦੇ ਉਤਪਾਦ।
- **ਹੈਲਪਰ**। LLMs ਟੈਕਸਟ ਸੰਖੇਪ ਕਰਨ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀ ਲੈਣ, ਰਿਜ਼ਿਊਮ ਵਰਗਾ ਟੈਕਸਟ ਬਣਾਉਣ ਆਦਿ ਵਿੱਚ ਵਧੀਆ ਹਨ।
- **ਕੋਡ ਅਸਿਸਟੈਂਟ**। ਮਾਡਲ ਦੇ ਅਨੁਸਾਰ, ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਅਸਿਸਟੈਂਟ ਵੀ ਬਣਾਉਣੀ ਹੈ ਜੋ ਤੁਹਾਨੂੰ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰੇ। ਉਦਾਹਰਨ ਵਜੋਂ, GitHub Copilot ਜਾਂ ChatGPT ਵਰਤ ਕੇ ਤੁਸੀਂ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਲੈ ਸਕਦੇ ਹੋ।

## ਸ਼ੁਰੂਆਤ ਕਿਵੇਂ ਕਰੀਏ?

ਤੁਹਾਨੂੰ LLM ਨਾਲ ਇੰਟੀਗਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪੈਂਦਾ ਹੈ, ਆਮ ਤੌਰ 'ਤੇ ਇਹ ਦੋ ਤਰੀਕੇ ਹੁੰਦੇ ਹਨ:

- API ਵਰਤੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣਾ ਪ੍ਰੋੰਪਟ ਲੈ ਕੇ ਵੈੱਬ ਰਿਕਵੈਸਟ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਵਾਪਸ ਲੈਂਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਵਰਤੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਾਂ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀਆਂ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀ/SDK

LLM ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਮਸ਼ਹੂਰ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਨੂੰ ਆਪਣੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰੋੰਪਟ ਭੇਜਣ ਵਿੱਚ ਆਸਾਨੀ ਦਿੰਦੀ ਹੈ।

ਇਸ ਤੋਂ ਉੱਚੇ ਪੱਧਰ 'ਤੇ ਕੰਮ ਕਰਨ ਵਾਲੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਵੀ ਹਨ:

- **Langchain**। Langchain ਮਸ਼ਹੂਰ ਹੈ ਅਤੇ Python ਲਈ ਸਹਾਇਕ ਹੈ।
- **Semantic Kernel**। Semantic Kernel Microsoft ਦੀ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python, ਅਤੇ Java ਲਈ ਸਹਾਇਕ ਹੈ।

## openai ਨਾਲ ਪਹਿਲਾ ਐਪ

ਆਓ ਵੇਖੀਏ ਕਿ ਪਹਿਲਾ ਐਪ ਕਿਵੇਂ ਬਣਾਇਆ ਜਾਵੇ, ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਚਾਹੀਦੀਆਂ ਹਨ, ਅਤੇ ਹੋਰ ਕੀ ਕੁਝ ਲੋੜੀਂਦਾ ਹੈ।

OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕਈ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ। ਤੁਸੀਂ ਕਈ ਪ੍ਰੋਗਰਾਮਿੰਗ ਭਾਸ਼ਾਵਾਂ ਵਰਤ ਸਕਦੇ ਹੋ, ਜਿਵੇਂ C#, Python, JavaScript, Java ਆਦਿ।  
ਅਸੀਂ `openai` Python ਲਾਇਬ੍ਰੇਰੀ ਚੁਣੀ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `pip` ਨਾਲ ਇੰਸਟਾਲ ਕਰਾਂਗੇ।

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Dev Container 'ਚ ਨਹੀਂ ਚਲਾ ਰਹੇ, ਤਾਂ ਤੁਹਾਨੂੰ ਆਪਣੇ ਕੰਪਿਊਟਰ 'ਤੇ [Python](https://www.python.org/) ਵੀ ਇੰਸਟਾਲ ਕਰਨਾ ਪਵੇਗਾ।

### ਸਰੋਤ ਬਣਾਓ ਅਤੇ ਆਪਣੀ API ਕੁੰਜੀ ਲੱਭੋ

ਜੇ ਤੁਸੀਂ ਪਹਿਲਾਂ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ ਇਹ ਕਦਮ ਕਰੋ:

- OpenAI 'ਤੇ ਖਾਤਾ ਬਣਾਓ <https://platform.openai.com/signup>।
- ਹੁਣ, ਆਪਣੀਆਂ API ਕੁੰਜੀਆਂ ਲਵੋ <https://platform.openai.com/api-keys>। 

>[!NOTE]
> ਆਪਣੀ API ਕੁੰਜੀ ਨੂੰ ਆਪਣੇ ਕੋਡ ਤੋਂ ਵੱਖ ਰੱਖਣਾ ਚੰਗਾ ਹੈ। ਤੁਸੀਂ ਇਹ ਵਾਤਾਵਰਣ ਚਲਾਉਣ ਵਾਲੀਆਂ ਵੈਰੀਏਬਲਾਂ ਨਾਲ ਕਰ ਸਕਦੇ ਹੋ।
> - ਵਾਤਾਵਰਣ ਵੈਰੀਏਬਲ `OPENAI_KEY` ਨੂੰ ਆਪਣੀ API ਕੁੰਜੀ ਨਾਲ .env ਫਾਈਲ ਵਿੱਚ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਪਹਿਲੇ ਅਭਿਆਸ ਕਰ ਲਏ ਹਨ, ਤਾਂ ਤੁਸੀਂ ਤਿਆਰ ਹੋ।
> - ਯਾਦ ਰੱਖੋ ਕਿ API ਕੁੰਜੀ ਸਿਰਫ ਇੱਕ ਵਾਰ ਹੀ ਮਿਲੇਗੀ। ਇਸ ਲਈ ਇਹ ਯਕੀਨੀ ਬਣਾਓ ਕਿ ਤੁਸੀਂ ਠੀਕ ਤਰੀਕੇ ਨਾਲ ਕਾਪੀ ਕਰ ਲਈ ਹੈ। ਜੇ ਇਹ ਕੰਮ ਨਾ ਕਰੇ, ਤਾਂ ਕੁੰਜੀ ਨੂੰ ਮਿਟਾ ਕੇ ਨਵੀਂ ਬਣਾਓ।

### OpenAI ਦੀ ਸੰਰਚਨਾ ਸੈੱਟ ਕਰੋ

ਜੇ ਤੁਸੀਂ OpenAI ਵਰਤ ਰਹੇ ਹੋ, ਤਾਂ ਸੰਰਚਨਾ ਇੰਝ ਸੈੱਟ ਕਰੋ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

ਉਪਰੋਕਤ ਕੋਡ ਵਿੱਚ ਅਸੀਂ ਇਹ ਸੈੱਟ ਕਰ ਰਹੇ ਹਾਂ:

- `api_key`, ਇਹ ਤੁਹਾਡੀ API ਕੁੰਜੀ ਹੈ ਜੋ OpenAI ਡੈਸ਼ਬੋਰਡ 'ਚ ਮਿਲਦੀ ਹੈ।
- `deployment`, ਇਹ ਤੁਹਾਡਾ GPT ਵਰਜਨ ਹੈ।

> [!NOTE]
> `os.environ` ਇੱਕ ਫੰਕਸ਼ਨ ਹੈ ਜੋ ਵਾਤਾਵਰਣ ਵੈਰੀਏਬਲਾਂ ਪੜ੍ਹਦਾ ਹੈ। ਤੁਸੀਂ ਇਸ ਨਾਲ `OPENAI_API_KEY` ਵਰਗੀਆਂ ਵੈਰੀਏਬਲਾਂ ਪੜ੍ਹ ਸਕਦੇ ਹੋ।

## ਟੈਕਸਟ ਜਨਰੇਟ ਕਰੋ

ਟੈਕਸਟ ਜਨਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ `chat.completion` ਕਲਾਸ ਵਰਤਣਾ ਹੈ। ਉਦਾਹਰਨ:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

ਉਪਰੋਕਤ ਕੋਡ ਵਿੱਚ, ਅਸੀਂ ਇੱਕ completion ਆਬਜੈਕਟ ਬਣਾਇਆ ਅਤੇ ਮਾਡਲ ਤੇ ਪ੍ਰੋੰਪਟ ਦਿੱਤਾ। ਫਿਰ ਅਸੀਂ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਪ੍ਰਿੰਟ ਕੀਤਾ।

### Chat completions

ਹੁਣ ਤੱਕ ਤੁਸੀਂ ਵੇਖਿਆ ਕਿ ਅਸੀਂ `Completion` ਵਰਤ ਕੇ ਟੈਕਸਟ ਜਨਰੇਟ ਕਰ ਰਹੇ ਹਾਂ। ਪਰ ਇੱਕ ਹੋਰ ਕਲਾਸ ਹੈ `ChatCompletion` ਜੋ ਚੈਟਬੋਟ ਲਈ ਵਧੀਆ ਹੈ। ਉਦਾਹਰਨ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )
```
