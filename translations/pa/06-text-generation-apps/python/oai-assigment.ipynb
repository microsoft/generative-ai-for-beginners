# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਸ ਬਣਾਓ

ਤੁਸੀਂ ਇਸ ਕਰੀਕੁਲਮ ਵਿੱਚ ਇਸ ਤੱਕ ਦੇਖਿਆ ਹੈ ਕਿ ਕੋਰ ਸੰਕਲਪ ਹਨ ਜਿਵੇਂ ਪ੍ਰੋੰਪਟ ਅਤੇ ਇੱਕ ਪੂਰੀ ਵਿਸ਼ੇਸ਼ਤਾ ਜੋ "ਪ੍ਰੋੰਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਕਹਿੰਦੀ ਹੈ। ਕਈ ਟੂਲਾਂ ਜਿਵੇਂ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ ਜਿਹਨਾਂ ਨਾਲ ਤੁਸੀਂ ਇੰਟਰੈਕਟ ਕਰ ਸਕਦੇ ਹੋ, ਤੁਹਾਡੇ ਕੰਮ ਨੂੰ ਮੁਕੰਮਲ ਕਰਨ ਲਈ ਪ੍ਰੋੰਪਟ ਵਰਤਣ ਦਾ ਸਹਾਰਾ ਦਿੰਦੇ ਹਨ।

ਤੁਹਾਡੇ ਲਈ ਐਪ ਵਿੱਚ ਐਸਾ ਅਨੁਭਵ ਜੋੜਨ ਲਈ, ਤੁਹਾਨੂੰ ਪ੍ਰੋੰਪਟ, ਪੂਰਨਤਾ ਅਤੇ ਕੰਮ ਕਰਨ ਲਈ ਇਕ ਲਾਈਬ੍ਰੇਰੀ ਚੁਣਨੇ ਵਰਗੇ ਸੰਕਲਪਾਂ ਨੂੰ ਸਮਝਣਾ ਜਰੂਰੀ ਹੈ। ਇਹੀ ਕੁਝ ਤੁਸੀਂ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖੋਗੇ।

## ਜਾਣ-ਪਹਚਾਣ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਈਬ੍ਰੇਰੀ ਅਤੇ ਇਸਦੇ ਮੁੱਖ ਸੰਕਲਪਾਂ ਬਾਰੇ ਸਿੱਖੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਨਾਲ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਂਦੇ ਹੋਏ।
- ਪ੍ਰੋੰਪਟ, ਤਾਪਮਾਨ, ਅਤੇ ਟੋਕਨ ਵਰਗੇ ਸੰਕਲਪਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਬਣਾਈ ਜਾਵੇ, ਇਹ ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲੱਛਣ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ 'ਤੇ, ਤੁਸੀਂ ਸਮਰੱਥ ਹੋਵੋਗੇ:

- ਸਮਝਾਉਣਾ ਕਿ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਨਾਲ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣਾ।
- ਆਪਣੇ ਐਪ ਨੂੰ ਵੱਧ ਜਾਂ ਘੱਟ ਟੋਕਨਾਂ ਦੀ ਵਰਤੋਂ ਕਰ ਲਈ ਕਨਫਿਗਰ ਕਰਨਾ, ਅਤੇ ਤਾਪਮਾਨ ਵੀ ਬਦਲ ਸਕਦੇ ਹੋ, ਤਾਂ ਜੋ ਵੱਖ-ਵੱਖ ਨਤੀਜੇ ਮਿਲਣ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਇਹ ਕਿਸੇ ਇੱਕ ਇੰਟਰਫੇਸ ਦੇ ਨਾਲ ਹੁੰਦੀ ਹੈ ਜਿਵੇਂ:

- ਕਮਾਂਡ-ਆਧਾਰਿਤ। ਕਂਸੋਲ ਐਪਸ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਇਹ ਕੰਮ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਣ ਲਈ, `git` ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪਸ ਵਿੱਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUIs) ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਸੁਮਿਤ ਕਰਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਆਦਿ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪਸ ਸੀਮਤ ਹਨ

ਇਨ੍ਹਾਂ ਦੀ ਤੁਲਨਾ ਇੱਕ ਕਮਾਂਡ-ਆਧਾਰਿਤ ਐਪ ਨਾਲ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ:

- **ਇਹ ਸੀਮਤ ਹੈ**। ਤੁਸੀਂ ਕਿਸੇ ਵੀ ਕਮਾਂਡ ਨੂੰ ਨਹੀਂ ਲਿਖ ਸਕਦੇ, ਸਿਰਫ ਉਹੀ ਜੋ ਐਪ ਸਮਰਥਿਤ ਕਰਦਾ ਹੈ।
- **ਭਾਸ਼ਾ-ਖ਼ਾਸ**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਸਮਰਥਿਤ ਕਰਦੇ ਹਨ, ਪਰ ਡਿਫਾਲਟ ਤੌਰ 'ਤੇ ਐਪ ਕਿਸੇ ਖ਼ਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣਾਇਆ ਜਾਂਦਾ ਹੈ, ਭਲੇ ਹੀ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾ ਸਮਰਥਿਤ ਵੀ ਜੋੜ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਾਂ ਦੇ ਫਾਇਦੇ

ਤਾਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੁੰਦਾ ਹੈ?

ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ, ਤੁਹਾਡੇ ਕੋਲ ਵੱਧ ਲਚਕੀਲਾਪਨ ਹੁੰਦਾ ਹੈ, ਤੁਸੀਂ ਕਿਸੇ ਨਿਸ਼ਚਿਤ ਕਮਾਂਡ ਜਾਂ ਭਾਸ਼ਾ ਵਿੱਚ ਸੀਮਿਤ ਨਹੀਂ ਹੋ। ਇਸ ਦੀ ਬਜਾਏ, ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਐਪ ਨਾਲ ਇੰਟਰੈਕਟ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਪਹਿਲਾਂ ਹੀ ਇੱਕ ਡੇਟਾ ਸਰੋਤ ਨਾਲ ਇੰਟਰੈਕਟ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੀ ਜਾਣਕਾਰੀ ਦੇ ਸੰਗ੍ਰਹਿ 'ਤੇ ਤਿਆਰ ਕੀਤਾ ਗਿਆ ਹੈ, ਜਦਕਿ ਪਰੰਪਰਾਗਤ ਐਪ ਸ਼ਾਇਦ ਸਿਰਫ ਡੇਟਾਬੇਸ ਵਿੱਚ ਹਨ ਉਹੀ ਸੀਮਿਤ ਰਹਿੰਦੇ ਹਨ।

### ਮੈਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂ ਸਕਦਾ ਹਾਂ?

ਬਹੁਤ ਕੁਝ ਬਣਾਇਆ ਜਾ ਸਕਦਾ ਹੈ। ਉਦਾਹਰਣ ਵਜੋਂ:

- **ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਤੁਹਾਡੇ ਕੰਪਨੀ ਅਤੇ ਇਸਦੇ ਉਤਪਾਦਾਂ ਬਾਰੇ ਪ੍ਰਸ਼ਨਾਂ ਦੇ ਜਵਾਬ ਦਿੰਦਾ ਹੈ, ਉਹ ਚੰਗਾ ਮਿਲਾਪ ਹੋ ਸਕਦਾ ਹੈ।
- **ਸਹਾਇਕ**। LLMs ਟੈਕਸਟ ਦਾ ਸੰਖੇਪ ਬਣਾਉਣ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀਆਂ ਪ੍ਰਾਪਤ ਕਰਨ, ਰਿਜ਼ੂਮੇ ਵਰਗਾ ਟੈਕਸਟ ਬਨਾਉਣ ਵਿੱਚ ਬਹੁਤ ਮਹਾਨ ਹਨ।
- **ਕੋਡ ਸਹਾਇਕ**। ਤੁਸੀਂ ਜਿਹੜਾ ਭਾਸ਼ਾ ਮਾਡਲ ਵਰਤਦੇ ਹੋ, ਉਸ ਦੇ ਅਨੁਸਾਰ ਕੋਡ ਸਹਾਇਕ ਬਣਾ ਸਕਦੇ ਹੋ ਜੋ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਣ ਲਈ, ਤੁਸੀਂ GitHub Copilot ਜਾਂ ChatGPT ਵਰਗੇ ਉਤਪਾਦਾਂ ਦੀ ਵਰਤੋਂ ਕਰ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਤੀਂਹ ਸ਼ੁਰੂ ਕਰਨ ਲਈ ਤੁਹਾਨੂੰ LLM ਨਾਲ ਇੰਟੀਗ੍ਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪਵੇਗਾ ਜੋ ਆਮ ਤੌਰ 'ਤੇ ਦੋ ਤਰੀਕਿਆਂ ਨਾਲ ਹੁੰਦਾ ਹੈ:

- API ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣਾ ਪ੍ਰੋੰਪਟ ਦੇ ਕੇ ਵੈੱਬ ਰਿਕਵੇਸਟ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀ API ਕਾਲਾਂ ਨੂੰ ਕੈਪਸੂਲ ਕਰਨ ਵਿੱਚ ਮਦਦ ਕਰਦੀਆਂ ਹਨ ਅਤੇ ਇਸ ਤਰ੍ਹਾਂ ਇਨ੍ਹਾ ਦੀ ਵਰਤੋਂ ਸੌਖੀ ਹੋ ਜਾਂਦੀ ਹੈ।

## ਲਾਇਬ੍ਰੇਰੀਜ਼/SDKs

ਕੁਝ ਪ੍ਰਸਿੱਧ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਹਨ ਜੋ LLMs ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰੋੰਪਟ ਭੇਜਣ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀ ਹੈ।

ਫਿਰ ਕੁਝ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਹਨ ਜੋ ਉੱਚ ਪੱਧਰ 'ਤੇ ਕੰਮ ਕਰਦੀਆਂ ਹਨ ਜਿਵੇਂ:

- **Langchain**। Langchain ਵੱਡੀ ਪਛਾਣ ਵਾਲੀ ਹੈ ਅਤੇ ਪਾਇਥਨ ਨੂੰ ਸਮਰਥਿਤ ਕਰਦੀ ਹੈ।
- **Semantic Kernel**। Semantic Kernel ਮਾਇਕਰੋਸਾਫਟ ਦੀ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python, ਅਤੇ Java ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਮਰਥਿਤ ਕਰਦੀ ਹੈ।

## openai ਨਾਲ ਪਹਿਲਾ ਐਪ

ਚਲੋ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਆਪਣਾ ਪਹਿਲਾ ਐਪ ਕਿਵੇਂ ਬਣਾ ਸਕਦੇ ਹਾਂ, ਸਾਨੂੰ ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਦੀ ਲੋੜ ਹੈ, ਕਿੰਨਾ ਕੁ ਚਾਹੀਦਾ ਹੈ ਅਤੇ ਹੋਰ।

### openai ਇਨਸਟਾਲ ਕਰੋ

  > [!NOTE] ਜੇ ਤੁਸੀਂ Codespaces ਜਾਂ Devcontainer ਵਿੱਚ ਇਹ ਨੋਟਬੁੱਕ ਚਲਾ ਰਹੇ ਹੋ ਤਾਂ ਇਹ ਕਦਮ ਜ਼ਰੂਰੀ ਨਹੀਂ ਹੈ


OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਇੰਟਰੈਕਟ ਕਰਨ ਲਈ ਕਈ ਲਾਇਬ੍ਰੇਰੀਆਂ ਉਪਲਬਧ ਹਨ। ਕਈ ਪ੍ਰੋਗ੍ਰਾਮਿੰਗ ਭਾਸ਼ਾਵਾਂ ਵੀ ਵਰਤੀ ਜਾ ਸਕਦੀਆਂ ਹਨ ਜਿਵੇਂ ਕਿ C#, Python, JavaScript, Java ਆਦਿ।  
ਅਸੀਂ `openai` ਪਾਇਥਨ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਚੋਣ ਕੀਤੀ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `pip` ਨਾਲ ਇਹ ਇਨਸਟਾਲ ਕਰਾਂਗੇ।

```bash
pip install openai
```

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Dev Container ਵਿੱਚ ਨਹੀਂ ਚਲਾ ਰਹੇ, ਤਾਂ ਤੁਹਾਨੂੰ ਆਪਣੇ ਮਸ਼ੀਨ 'ਤੇ [Python](https://www.python.org/) ਵੀ ਇਨਸਟਾਲ ਕਰਨਾ ਪਵੇਗਾ।

### ਇਕ ਸਰੋਤ ਬਣਾਓ ਅਤੇ ਆਪਣਾ API ਕੀ ਲੱਭੋ

ਜੇ ਤੁਸੀਂ ਪਹਿਲਾਂ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ ਤੁਹਾਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਕਦਮ ਕਰਨੇ ਪੈਂਦੇ ਹਨ:

- OpenAI ਤੇ ਇਕ ਅਕਾਊਂਟ ਬਣਾਓ <https://platform.openai.com/signup>.
- ਹੁਣ, ਆਪਣੀਆਂ API ਕੁੰਜੀਆਂ ਪ੍ਰਾਪਤ ਕਰੋ <https://platform.openai.com/api-keys>. 

>[!NOTE]
> ਆਪਣੀ API ਕੀ ਆਪਣੇ ਕੋਡ ਤੋਂ ਵੱਖਰੀ ਰੱਖਣ ਵਾਜਬ ਹੈ। ਤੁਸੀਂ ਇਹ ਤਰੀਕਾ ਵਰਤ ਕੇ ਕਰ ਸਕਦੇ ਹੋ - ਪਰਿਵੇਸ਼ ਵੈਰੀਏਬਲ।
> - ਆਪਣੇ .env ਫਾਈਲ ਵਿੱਚ `OPENAI_KEY` ਵੈਰੀਏਬਲ ਨੂੰ ਆਪਣੀ API ਕੀ ਨਾਲ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਇਸ ਕੋर्स ਦੀਆਂ ਪਿਛਲੀਆਂ ਕਸਰਤਾਂ ਮੁਕੰਮਲ ਕਰ ਲਈਆਂ ਹਨ ਤਾਂ ਤੁਹਾਡਾ ਸੈਟਅੱਪ ਹੋ ਚੁੱਕਾ ਹੈ।
> - ਇਹ ਜ਼ਰੂਰੀ ਹੈ ਕਿ API ਕਿੜਾਰੀਵਾਰੀ ਕੇਵਲ ਇੱਕ ਵਾਰੀ ਹੀ ਐਕਸੈਸ ਕੀਤੀ ਜਾਵੇਗੀ। ਇਸ ਲਈ, ਇਹ ਯਕੀਨੀ ਬਣਾਓ ਕਿ ਇਹ ਸਹੀ ਤਰੀਕੇ ਨਾਲ ਨਕਲ ਕੀਤੀ ਗਈ ਹੈ। ਜੇ ਇਹ ਕੰਮ ਨਹੀਂ ਕਰਦੀ ਤਾਂ ਇਹਥੋਂ ਕੀ ਨੂੰ ਹਟਾ ਕੇ ਨਵੀਂ ਕੀ ਬਣਾਉਣਾ ਚਾਹੀਦਾ ਹੈ।


### OpenAI ਕਨਫਿਗਰੇਸ਼ਨ ਸੈਟਅੱਪ

ਜੇ ਤੁਸੀਂ OpenAI ਵਰਤ ਰਹੇ ਹੋ, ਤਾਂ ਕਨਫਿਗਰੇਸ਼ਨ ਸੈਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਇਹ ਹੈ:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

ਉਪਰ ਅਸੀਂ ਇਹ ਸੈਟ ਕਰ ਰਹੇ ਹਾਂ:

- `api_key`, ਇਹ ਤੁਹਾਡੀ API ਕੀ ਹੈ ਜੋ OpenAI ਡੈਸ਼ਬੋਰਡ ਵਿੱਚ ਮਿਲਦੀ ਹੈ।
- `deployment`, ਇਹ ਤੁਹਾਡਾ GPT ਵਰਜ਼ਨ ਹੈ।

> [!NOTE]

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- ਇਤਿਹਾਸ ਬੋਟ, ਇਤਿਹਾਸ ਨੂੰ ਜੀਵੰਤ ਬਣਾ ਦਿਓ, ਬੋਟ ਨੂੰ ਕਹੋ ਕਿ ਉਹ ਕਿਸੇ ਇਤਿਹਾਸਕ ਪ੍ਰਸਿੱਧ ਕਿਰਦਾਰ ਦੀ ਭੂਮਿਕਾ ਨਿਭਾਏ ਅਤੇ ਉਸ ਦੀ ਜ਼ਿੰਦਗੀ ਅਤੇ ਸਮਿਆਂ ਬਾਰੇ ਸਵਾਲ ਪੁੱਛੋ।
