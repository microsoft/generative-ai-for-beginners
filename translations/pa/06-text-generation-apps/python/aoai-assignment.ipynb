# ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪਸ ਬਣਾਓ

ਤੁਹਾਡੇ ਕੋਲ ਹੁਣ ਤੱਕ ਇਸ ਕੋਰਸ ਵਿੱਚ ਇਹ ਦੇਖਿਆ ਹੈ ਕਿ ਬੁਨਿਆਦੀ ਧਾਰਣਾਵਾਂ ਜਿਵੇਂ ਕਿ ਪ੍ਰੌਮਟਸ ਅਤੇ ਇੱਕ ਪੂਰਾ ਵਿਭਾਗ "ਪ੍ਰੌਮਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਵੀ ਹੈ। ਬਹੁਤ ਸਾਰੇ ਟੂਲ ਜਿਵੇਂ ਕਿ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ ਹਨ ਜੋ ਤੁਹਾਨੂੰ ਕੁਝ ਕਰਨ ਲਈ ਪ੍ਰੌਮਟਸ ਦੀ ਵਰਤੋਂ ਕਰਨ ਦੀ ਸਹਾਇਤਾ ਕਰਦੇ ਹਨ।

ਕਿਸੇ ਐਪ ਵਿੱਚ ਅਜਿਹੇ ਤਜਰਬੇ ਨੂੰ ਸ਼ਾਮਿਲ ਕਰਨ ਲਈ, ਤੁਹਾਨੂੰ ਪ੍ਰੌਮਟਸ, ਕਮਪਲੀਸ਼ਨਜ਼ ਜਿਹੀਆਂ ਧਾਰਣਾਵਾਂ ਨੂੰ ਸਮਝਣਾ ਪੈਂਦਾ ਹੈ ਅਤੇ ਕਿਸੇ ਲਾਇਬ੍ਰੇਰੀ ਨੂੰ ਚੁਣਨਾ ਪੈਂਦਾ ਹੈ ਜਿਸ ਨਾਲ ਕੰਮ ਕਰਨਾ ਹੈ। ਇਹੀ ਤੁਹਾਨੂੰ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖਣ ਨੂੰ ਮਿਲੇਗਾ।

## ਪਰਿਚਯ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸ ਦੀਆਂ ਬੁਨਿਆਦੀ ਧਾਰਣਾਵਾਂ ਬਾਰੇ ਸਿੱਖੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਬਣਾਉਗੇ।
- ਪ੍ਰੌਮਟ, ਟੈਮਪਰੇਚਰ, ਅਤੇ ਟੋਕਨਜ਼ ਵਰਗੀਆਂ ਧਾਰṇਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਬਣਾਈ ਜਾਵੇ, ਇਹ ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜੇ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਤਕ ਤੁਸੀਂ ਸਮਰੱਥ ਹੋਵੋਗੇ:

- ਇਕ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦਾ ਹੈ, ਸਮਝਾਓ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਬਣਾਓ।
- ਆਪਣੇ ਐਪ ਨੂੰ ਵੱਧ ਜਾਂ ਘੱਟ ਟੋਕਨਜ਼ ਦੀ ਵਰਤੋਂ ਲਈ ਕਨਫਿਗਰ ਕਰੋ ਅਤੇ ਨਤੀਜੇ ਵਿੱਚ ਵੱਖ-ਵੱਖ ਤਾਪਮਾਨ ਬਦਲੋ।

## ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਉਸਦਾ ਕੋਈ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ ਜਿਵੇਂ ਕਿ ਹੇਠ ਲਿਖਿਆ ਹੈ:

- ਕਮਾਂਡ-ਅਧਾਰਿਤ. ਕਨਸੋਲ ਐਪ ਉਹ ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਉਹ ਕੰਮ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI). ਕੁਝ ਐਪ ਵਿਚ ਗ੍ਰਾਫਿਕ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUIs) ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਇਨਪੁੱਟ ਕਰਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਅਤੇ ਹੋਰ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪ ਸੀਮਿਤ ਹਨ

ਇਸਦੀ ਤੁਲਨਾ ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਨਾਲ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ:

- **ਇਹ ਸੀਮਿਤ ਹੈ**। ਤੁਸੀਂ ਕਿਵੇਂ ਵੀ ਕਮਾਂਡ ਟਾਈਪ ਨਹੀਂ ਕਰ ਸਕਦੇ, ਸਿਰਫ ਉਹੀ ਜੋ ਐਪ ਸਹਾਇਤਾ ਕਰਦਾ ਹੈ।
- **ਭਾਸ਼ਾ-ਵਿਸ਼ੇਸ਼**। ਕੁਝ ਐਪ ਬਹੁਤ ਸਾਰੀਆਂ ਭਾਸ਼ਾਵਾਂ ਸਹਿਯੋਗ ਕਰਦੇ ਹਨ, ਪਰ ਆਮ ਤੌਰ 'ਤੇ ਐਪ ਇੱਕ ਖ਼ਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣਾਇਆ ਜਾਂਦਾ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਦਾ ਸਹਿਯੋਗ ਜੋੜ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਦੇ ਲਾਭ

ਤਾਂ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਕਿਵੇਂ ਵੱਖਰਾ ਹੁੰਦਾ ਹੈ?

ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਵੱਧ ਲਚਕੀਲਾਪਨ ਹੁੰਦੀ ਹੈ, ਤੁਸੀਂ ਕਿਸੇ ਖ਼ਾਸ ਕਮਾਂਡ ਜਾਂ ਭਾਸ਼ਾ 'ਤੇ ਸੀਮਿਤ ਨਹੀਂ ਹੁੰਦੇ। ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਐਪ ਨਾਲ ਗੱਲਬਾਤ ਕਰ ਸਕਦੇ ਹੋ। ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਪਹਿਲਾਂ ਹੀ ਬਹੁਤ ਸਾਰੇ ਜਾਣਕਾਰੀ ਵਾਲੇ ਡਾਟਾ ਸੋਰਸ ਨਾਲ ਇੰਟਰאַקט ਕਰ ਰਹੇ ਹੋ, ਜਿੱਥੇ ਕਿ ਰਵਾਇਤੀ ਐਪ ਡਾਟਾਬੇਸ ਵਿਚ ਹਦਬੰਦ ਹੋ ਸਕਦਾ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜਨੇਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂਦਾ ਹਾਂ?

ਤੁਸੀਂ ਕਈ ਚੀਜ਼ਾਂ ਬਣਾਉਂਦੇ ਹੋ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਵਜੋਂ:

- **ਇੱਕ ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਤੁਹਾਡੇ ਕੰਪਨੀ ਅਤੇ ਇਸਦੇ ਉਤਪਾਦਾਂ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇ ਸਕਦਾ ਹੈ, ਇੱਕ ਵਧੀਆ ਮਿਸ਼ਾਲ ਹੋ ਸਕਦਾ ਹੈ।
- **ਸਹਾਇਕ**। LLMs ਟੈਕਸਟ ਸਾਰਾਂਸ਼ ਬਣਾਉਣ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀ ਪ੍ਰਾਪਤ ਕਰਨ, ਰੈਜ਼ਿਊਮੇ ਵੀਲੇਖਣ ਅਤੇ ਹੋਰ ਕੰਮਾਂ ਵਿੱਚ ਬਹੁਤ ਖੂਬ ਹਨ।
- **ਕੋਡ ਸਹਾਇਕ**। ਤੁਸੀਂ ਜੋ ਭਾਸ਼ਾ ਮਾਡਲ ਵਰਤ ਰਹੇ ਹੋ, ਉਸਦੇ ਮੁਤਾਬਕ ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਸਹਾਇਕ ਬਣਾ ਸਕਦੇ ਹੋ ਜੋ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਸਹਾਇਤਾ ਕਰਦਾ ਹੈ। ਉਦਾਹਰਨ ਵਜੋਂ, ਤੁਸੀਂ GitHub Copilot ਅਤੇ ChatGPT ਵਰਗੇ ਉਤਪਾਦਾਂ ਦੀ ਵਰਤੋਂ ਕਰ ਸਕਦੇ ਹੋ ਕੋਡ ਲਿਖਣ ਲਈ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਚੰਗਾ, ਤੁਹਾਨੂੰ ਕਈ ਵਾਰ LLM ਨਾਲ ਇੰਟੀਗ੍ਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪੈਂਦਾ ਹੈ ਜਿਸ ਲਈ ਆਮ ਤੌਰ 'ਤੇ ਦੋ ਤਰੀਕੇ ਹਨ:

- API ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣੇ ਪ੍ਰੌਮਟ ਨਾਲ ਵੈੱਬ ਰਿਕਵੇਸਟ ਬਨਾਉਂਦੇ ਹੋ ਅਤੇ ਜਨੇਰੇਟ ਕੀਤਾ ਹੋਇਆ ਟੈਕਸਟ ਮੁੜ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਜ਼ ਨੂੰ ਘੇਰ ਲੈਂਦੀਆਂ ਹਨ ਅਤੇ ਇਹਨਾਂ ਨੂੰ ਵਰਤਣਾ ਆਸਾਨ ਬਣਾਉਂਦੀਆਂ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀਆਂ/SDKs

LLM ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਜਾਣੇ ਮਾਣੇ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ ਕਿ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਡੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰੌਮਟ ਭੇਜਣ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀ ਹੈ।

ਫਿਰ ਕੁਝ ਇਸ ਤੋਂ ਉੱਚਾ ਸਤਰ 'ਤੇ ਕੰਮ ਕਰਨ ਵਾਲੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **Langchain**। Langchain ਵਧੀਆ ਜਾਣੀ ਜਾਂਦੀ ਹੈ ਅਤੇ ਪਾਇਥਨ ਸਹਿਯੋਗ ਕਰਦੀ ਹੈ।
- **Semantic Kernel**। Semantic Kernel Microsoft ਵੱਲੋਂ ਇੱਕ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python, ਅਤੇ Java ਭਾਸ਼ਾਵਾਂ ਸਹਿਯੋਗ ਕਰਦੀ ਹੈ।

## openai ਦੀ ਵਰਤੋਂ ਨਾਲ ਪਹਿਲਾ ਐਪ

ਆਓ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਆਪਣਾ ਪਹਿਲਾ ਐਪ ਕਿਵੇਂ ਬਣਾਉਣਗੇ, ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਦੀ ਲੋੜ ਹੈ, ਕਿੰਨੀ ਜ਼ਰੂਰਤ ਹੁੰਦੀ ਹੈ ਆਦਿ।

### openai ਇੰਸਟਾਲ ਕਰੋ

  > [!NOTE] ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Devcontainer ਵਿੱਚ ਚਲਾ ਰਹੇ ਹੋ ਤਾਂ ਇਹ ਕਦਮ ਜ਼ਰੂਰੀ ਨਹੀਂ ਹੈ


OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਇੰਟਰੈਕਟ ਕਰਨ ਲਈ ਬਹੁਤ ਸਾਰੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ। ਇਸਦੇ ਨਾਲ ਹੀ ਤੁਸੀਂ ਕਈ ਪ੍ਰੋਗ੍ਰਾਮਿੰਗ ਭਾਸ਼ਾਵਾਂ ਜਿਵੇਂ ਕਿ C#, Python, JavaScript, Java ਆਦਿ ਦੀ ਵਰਤੋਂ ਕਰ ਸਕਦੇ ਹੋ।
ਅਸੀਂ `openai` ਪਾਇਥਨ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣੀ ਹੈ, ਤਾਂ ਅਸੀਂ ਇਸਨੂੰ ਇੰਸਟਾਲ ਕਰਨ ਲਈ `pip` ਦੀ ਵਰਤੋਂ ਕਰਾਂਗੇ।

```bash
pip install openai
```

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Dev Container ਵਿੱਚ ਨਹੀਂ ਚਲਾ ਰਹੇ ਹੋ ਤਾਂ ਤੁਹਾਨੂੰ ਆਪਣੇ ਮਸ਼ੀਨ ਤੇ [Python](https://www.python.org/) ਵੀ ਇੰਸਟਾਲ ਕਰਨਾ ਪਵੇਗਾ।

### ਇੱਕ ਰਿਸੋਰਸ ਬਣਾਓ

ਜੇ ਤੁਸੀਂ ਪਹਿਲਾਂ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ ਹੇਠ ਲਿਖੇ ਕਦਮ ਕਰੋ:

### API ਕੀ ਅਤੇ ਐਂਡਪੌਇੰਟ ਲੱਭੋ

ਇਸ ਸਮੇਂ, ਤੁਹਾਨੂੰ ਆਪਣੇ `openai` ਲਾਇਬ੍ਰੇਰੀ ਨੂੰ ਦੱਸਣਾ ਪਵੇਗਾ ਕਿ ਕਿਹੜੀ API ਕੀ ਵਰਤੀ ਜਾਵੇ। ਆਪਣੀ API ਕੀ ਲੱਭਣ ਲਈ, ਆਪਣੇ Azure OpenAI ਰਿਸੋਰਸ ਦੇ "Keys and Endpoint" ਸੈਕਸ਼ਨ 'ਚ ਜਾਓ ਅਤੇ "Key 1" ਦਾ ਮੁੱਲ ਕਾਪੀ ਕਰੋ।

  ![Azure ਪੋਰਟਲ ਵਿੱਚ Keys and Endpoint ਰਿਸੋਰਸ ਬਲੇਡ](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ਹੁਣ ਜਦੋਂ ਤੁਹਾਡੇ ਕੋਲ ਇਹ ਜਾਣਕਾਰੀ ਹੈ, ਆਓ ਲਾਇਬ੍ਰੇਰੀਆਂ ਨੂੰ ਇਸਦੀ ਵਰਤੋਂ ਕਰਨ ਲਈ ਕਹੀਏ।

> [!NOTE]
> ਆਪਣੀ API ਕੀ ਨੂੰ ਕੋਡ ਤੋਂ ਵੱਖਰਾ ਰੱਖਣਾ ਲਾਭਦਾਇਕ ਹੈ। ਤੁਸੀਂ ਇਹ ਪੜਦਾਰ ਵਾਤਾਵਰਣ ਚਲਾਂਵਾਂ (environment variables) ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਕਰ ਸਕਦੇ ਹੋ।
> - ਆਪਣੀ .env ਫਾਇਲ ਵਿੱਚ ਵਾਤਾਵਰਣ ਚਲ `AZURE_OPENAI_API_KEY` ਨੂੰ ਆਪਣੀ API ਕੀ ਤੇ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਇਸ ਕੋਰਸ ਦੀਆਂ ਪਹਿਲਾਂ ਦੀਆਂ ਅਭਿਆਸਾਂ ਕਰ ਲਈਆਂ ਹਨ, ਤਾਂ ਤੁਸੀਂ ਤਿਆਰ ਹੋ।


### Azure ਸੰਰਚਨਾ ਸੈੱਟਅੱਪ ਕਰੋ

ਜੇ ਤੁਸੀਂ Azure OpenAI ਦੀ ਵਰਤੋਂ ਕਰ ਰਹੇ ਹੋ, ਤਾਂ ਇੱਥੇ ਤੁਹਾਡੇ ਲਈ ਸੰਰਚਨਾ ਸੈੱਟਅੱਪ ਕਰਨ ਦਾ ਤਰੀਕਾ ਹੈ। Responses API Azure OpenAI **v1 endpoint** ਤੋਂ ਸੇਵਾ ਦਿੱਤਾ ਜਾਂਦਾ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `OpenAI` ਕਲਾਇੰਟ ਨੂੰ `<your-endpoint>/openai/v1/` 'ਤੇ ਪੋਇੰਟ ਕਰਦੇ ਹਾਂ:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ਉਪਰ ਅਸੀਂ ਹੇਠਾਂ ਦਿੱਤੀਆਂ ਚੀਜ਼ਾਂ ਸੈਟ ਕਰ ਰਹੇ ਹਾਂ:

- `api_key`, ਇਹ ਤੁਹਾਡੀ Azure ਪੋਰਟਲ ਤੋਂ ਮਿਲੀ API ਕੀ ਹੈ।
- `base_url`, ਇਹ ਤੁਹਾਡਾ Azure OpenAI ਐਂਡਪੌਇੰਟ ਹੈ ਜਿਸ ਨੂੰ `/openai/v1/` ਨਾਲ ਜੋੜਿਆ ਗਿਆ ਹੈ। ਤੁਸੀਂ ਇਹ ਐਂਡਪੌਇੰਟ Azure ਪੋਰਟਲ ਵਿੱਚ ਆਪਣੀ API ਕੀ ਦੇ ਕੇਰੇਂ (ਨਜ਼ਦੀਕ) ਲੱਭ ਸਕਦੇ ਹੋ। v1 ਐਂਡਪੌਇੰਟ ਦੀ ਵਰਤੋਂ ਨਾਲ ਤੁਹਾਨੂੰ `api_version` ਦੇਣ ਦੀ ਲੋੜ ਨਹੀਂ ਰਹਿੰਦੀ।
- `deployment`, ਇਹ Foundry ਪੋਰਟਲ ਵਿੱਚ ਤੁਸੀਂ ਬਣਾਇਆ ਮਾਡਲ ਡਿਪਲয়ਮੈਂਟ ਦਾ ਨਾਮ ਹੈ।

> [!NOTE]
> `os.environ` ਇੱਕ ਮੈਪਿੰਗ ਹੈ ਜੋ ਵਾਤਾਵਰਣ ਚਲਾਂ ਨੂੰ ਪੜ੍ਹਦਾ ਹੈ। ਤੁਸੀਂ ਇਸਨੂੰ ਵਰਤ ਕੇ ਵਾਤਾਵਰਣ ਚਲਾਂ ਜਿਵੇਂ `AZURE_OPENAI_API_KEY` ਅਤੇ `AZURE_OPENAI_ENDPOINT` ਪੜ੍ਹ ਸਕਦੇ ਹੋ।

## ਟੈਕਸਟ ਜਨੇਰੇਟ ਕਰੋ

ਟੈਕਸਟ ਜਨੇਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਹੈ `responses.create` ਮੇਥਡ ਦੀ ਵਰਤੋਂ ਕਰਨੀ। ਹੇਠਾਂ ਇੱਕ ਉਦਾਹਰਨ ਦਿੱਤੀ ਗਈ ਹੈ:


### ਚੈਟ ਪ੍ਰਤੀਕਿਰਿਆਵਾਂ
