# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ

ਤੁਸੀਂ ਹੁਣ ਤੱਕ ਇਸ ਪਾਠਕ੍ਰਮ ਰਾਹੀਂ ਦੇਖਿਆ ਹੈ ਕਿ ਕੁਝ ਮੁੱਖ ਧਾਰਾਵਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰੌਂਪਟ ਅਤੇ ਇੱਥੋਂ ਤੱਕ ਕਿ "ਪ੍ਰੌਂਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਨਾਂ ਦੀ ਪੂਰੀ ਵਿਧਾ ਵੀ ਹੈ। ਕਈ ਟੂਲ ਜਿਵੇਂ ChatGPT, Office 365, Microsoft Power Platform ਆਦਿ ਤੁਹਾਨੂੰ ਪ੍ਰੌਂਪਟ ਵਰਤ ਕੇ ਕੁਝ ਹਾਸਲ ਕਰਨ ਵਿੱਚ ਮਦਦ ਕਰਦੇ ਹਨ।

ਜੇ ਤੁਸੀਂ ਆਪਣੀ ਐਪ ਵਿੱਚ ਇਹ ਤਜਰਬਾ ਸ਼ਾਮਲ ਕਰਨਾ ਚਾਹੁੰਦੇ ਹੋ, ਤਾਂ ਤੁਹਾਨੂੰ ਪ੍ਰੌਂਪਟ, ਕੰਪਲੀਸ਼ਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਅਤੇ ਵਰਤਣ ਲਈ ਲਾਇਬ੍ਰੇਰੀ ਚੁਣਣੀ ਪਵੇਗੀ। ਇਹੀ ਗੱਲ ਤੁਸੀਂ ਇਸ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖੋਗੇ।

## ਜਾਣ-ਪਛਾਣ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸ ਦੀਆਂ ਮੁੱਖ ਧਾਰਾਵਾਂ ਬਾਰੇ ਜਾਣੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਿੱਖੋਗੇ।
- ਪ੍ਰੌਂਪਟ, ਟੈਮਪਰੇਚਰ ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਧਾਰਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ 'ਤੇ, ਤੁਸੀਂ ਸਮਝ ਸਕੋਗੇ:

- ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣੀ।
- ਆਪਣੀ ਐਪ ਵਿੱਚ ਟੋਕਨ ਦੀ ਗਿਣਤੀ ਵਧਾਉਣ-ਘਟਾਉਣ ਅਤੇ ਟੈਮਪਰੇਚਰ ਬਦਲਣ ਦੀ ਸੰਰਚਨਾ ਕਰ ਸਕਦੇ ਹੋ, ਤਾਂ ਜੋ ਨਤੀਜਾ ਵੱਖ-ਵੱਖ ਆ ਸਕੇ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਕੋਈ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਉਸਦਾ ਕੁਝ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ, ਜਿਵੇਂ:

- ਕਮਾਂਡ-ਅਧਾਰਤ। ਕਨਸੋਲ ਐਪ ਆਮ ਤੌਰ 'ਤੇ ਉਹ ਐਪ ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ ਅਤੇ ਉਹ ਕੰਮ ਕਰ ਦਿੰਦੇ ਹਨ। ਉਦਾਹਰਨ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪਾਂ ਵਿੱਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUI) ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਲਿਖਦੇ ਹੋ, ਚੋਣ ਕਰਦੇ ਹੋ ਆਦਿ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪਾਂ ਦੀਆਂ ਹੱਦਾਂ

ਇਸਨੂੰ ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਤ ਐਪ ਨਾਲ ਤੁਲਨਾ ਕਰੋ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਲਿਖਦੇ ਹੋ:

- **ਇਹ ਸੀਮਤ ਹੈ**। ਤੁਸੀਂ ਹਰ ਕਿਸਮ ਦੀ ਕਮਾਂਡ ਨਹੀਂ ਲਿਖ ਸਕਦੇ, ਸਿਰਫ ਉਹੀ ਜੋ ਐਪ ਸਹਾਇਕ ਕਰਦੀ ਹੈ।
- **ਭਾਸ਼ਾ-ਨਿਰਧਾਰਤ**। ਕੁਝ ਐਪ ਕਈ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਕ ਕਰਦੀਆਂ ਹਨ, ਪਰ ਆਮ ਤੌਰ 'ਤੇ ਐਪ ਇੱਕ ਖਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣਾਈ ਜਾਂਦੀ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਜੋੜ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਾਂ ਦੇ ਫਾਇਦੇ

ਤਾਂ ਫਿਰ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵੱਖਰੀ ਕਿਵੇਂ ਹੈ?

ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਹੋਰ ਲਚਕ ਹੁੰਦੀ ਹੈ, ਤੁਸੀਂ ਸਿਰਫ਼ ਕੁਝ ਕਮਾਂਡਾਂ ਜਾਂ ਖਾਸ ਭਾਸ਼ਾ ਤੱਕ ਸੀਮਤ ਨਹੀਂ ਹੋ। ਇਸ ਦੀ ਬਜਾਏ, ਤੁਸੀਂ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਵਿੱਚ ਐਪ ਨਾਲ ਗੱਲ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਇੱਕ ਐਸੇ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਗੱਲ ਕਰ ਰਹੇ ਹੋ ਜੋ ਬਹੁਤ ਵੱਡੀ ਜਾਣਕਾਰੀ 'ਤੇ ਟਰੇਨ ਹੋਇਆ ਹੈ, ਜਦਕਿ ਰਵਾਇਤੀ ਐਪ ਸਿਰਫ਼ ਡਾਟਾਬੇਸ ਤੱਕ ਸੀਮਤ ਹੋ ਸਕਦੀ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾਉਂ ਸਕਦਾ ਹਾਂ?

ਤੁਸੀਂ ਕਈ ਚੀਜ਼ਾਂ ਬਣਾਉਣੀਆਂ। ਉਦਾਹਰਨ ਵਜੋਂ:

- **ਚੈਟਬੋਟ**। ਚੈਟਬੋਟ ਜੋ ਕਿਸੇ ਵਿਸ਼ੇ, ਜਿਵੇਂ ਤੁਹਾਡੀ ਕੰਪਨੀ ਜਾਂ ਉਤਪਾਦਾਂ ਬਾਰੇ ਸਵਾਲਾਂ ਦੇ ਜਵਾਬ ਦੇਵੇ, ਵਧੀਆ ਚੋਣ ਹੋ ਸਕਦੀ ਹੈ।
- **ਹੈਲਪਰ**। LLMs ਟੈਕਸਟ ਸੰਖੇਪ ਕਰਨ, ਟੈਕਸਟ ਤੋਂ ਜਾਣਕਾਰੀ ਲੈਣ, ਰਿਜ਼ਿਊਮੇ ਵਰਗਾ ਟੈਕਸਟ ਬਣਾਉਣ ਆਦਿ ਵਿੱਚ ਬਹੁਤ ਵਧੀਆ ਹਨ।
- **ਕੋਡ ਅਸਿਸਟੈਂਟ**। ਤੁਸੀਂ ਜਿਸ ਭਾਸ਼ਾ ਮਾਡਲ ਦੀ ਵਰਤੋਂ ਕਰਦੇ ਹੋ, ਉਸ ਅਨੁਸਾਰ ਤੁਸੀਂ ਇੱਕ ਕੋਡ ਅਸਿਸਟੈਂਟ ਵੀ ਬਣਾਉਣੀ ਜੋ ਤੁਹਾਨੂੰ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰੇ। ਉਦਾਹਰਨ ਵਜੋਂ, ਤੁਸੀਂ GitHub Copilot ਜਾਂ ChatGPT ਵਰਗੇ ਉਤਪਾਦ ਵਰਤ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰਾਂ?

ਤੁਹਾਨੂੰ LLM ਨਾਲ ਇੰਟੀਗਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪਵੇਗਾ, ਆਮ ਤੌਰ 'ਤੇ ਇਹ ਦੋ ਤਰੀਕਿਆਂ 'ਚੋਂ ਇੱਕ ਹੁੰਦਾ ਹੈ:

- API ਵਰਤੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣਾ ਪ੍ਰੌਂਪਟ ਲੈ ਕੇ ਵੈੱਬ ਰਿਕਵੈਸਟ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਵਾਪਸ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਮਿਲਦਾ ਹੈ।
- ਲਾਇਬ੍ਰੇਰੀ ਵਰਤੋ। ਲਾਇਬ੍ਰੇਰੀਆਂ API ਕਾਲਾਂ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀਆਂ ਹਨ।

## ਲਾਇਬ੍ਰੇਰੀਆਂ/SDKs

LLMs ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਮਸ਼ਹੂਰ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ:

- **openai**, ਇਹ ਲਾਇਬ੍ਰੇਰੀ ਤੁਹਾਨੂੰ ਆਪਣੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨ ਅਤੇ ਪ੍ਰੌਂਪਟ ਭੇਜਣ ਵਿੱਚ ਆਸਾਨੀ ਦਿੰਦੀ ਹੈ।

ਫਿਰ ਕੁਝ ਹੋਰ ਉੱਚ ਪੱਧਰ ਦੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ:

- **Langchain**। Langchain ਮਸ਼ਹੂਰ ਹੈ ਅਤੇ Python ਲਈ ਸਹਾਇਕ ਹੈ।
- **Semantic Kernel**। Semantic Kernel Microsoft ਵੱਲੋਂ ਬਣਾਈ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, Python ਅਤੇ Java ਲਈ ਉਪਲਬਧ ਹੈ।

## openai ਨਾਲ ਪਹਿਲੀ ਐਪ

ਆਓ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਆਪਣੀ ਪਹਿਲੀ ਐਪ ਕਿਵੇਂ ਬਣਾਈਏ, ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਚਾਹੀਦੀਆਂ ਹਨ, ਕਿੰਨਾ ਕੰਮ ਕਰਨਾ ਪੈਂਦਾ ਹੈ ਆਦਿ।

### openai ਇੰਸਟਾਲ ਕਰੋ

  > [!NOTE] ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Devcontainer 'ਚ ਚਲਾ ਰਹੇ ਹੋ ਤਾਂ ਇਹ ਕਦਮ ਲਾਜ਼ਮੀ ਨਹੀਂ।

OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕਈ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ। ਤੁਸੀਂ ਕਈ ਭਾਸ਼ਾਵਾਂ ਵਰਤ ਸਕਦੇ ਹੋ ਜਿਵੇਂ C#, Python, JavaScript, Java ਆਦਿ।  
ਅਸੀਂ `openai` Python ਲਾਇਬ੍ਰੇਰੀ ਚੁਣੀ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `pip` ਨਾਲ ਇਸਨੂੰ ਇੰਸਟਾਲ ਕਰਾਂਗੇ।

```bash
pip install openai
```

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Dev Container ਵਿੱਚ ਨਹੀਂ ਚਲਾ ਰਹੇ, ਤਾਂ ਤੁਹਾਨੂੰ ਆਪਣੇ ਕੰਪਿਊਟਰ 'ਤੇ [Python](https://www.python.org/) ਵੀ ਇੰਸਟਾਲ ਕਰਨਾ ਪਵੇਗਾ।

### ਸਰੋਤ ਬਣਾਓ

ਜੇ ਤੁਸੀਂ ਪਹਿਲਾਂ ਨਹੀਂ ਕੀਤਾ, ਤਾਂ ਇਹ ਕਦਮ ਕਰੋ:

- Azure 'ਤੇ ਖਾਤਾ ਬਣਾਓ <https://azure.microsoft.com/free/>।
- Azure OpenAI ਦੀ ਪਹੁੰਚ ਲਵੋ। <https://learn.microsoft.com/azure/ai-services/openai/overview#how-do-i-get-access-to-azure-openai> 'ਤੇ ਜਾ ਕੇ ਪਹੁੰਚ ਲਈ ਅਰਜ਼ੀ ਦਿਓ।

  > [!NOTE]
  > ਲਿਖਣ ਦੇ ਸਮੇਂ, ਤੁਹਾਨੂੰ Azure OpenAI ਲਈ ਅਰਜ਼ੀ ਦੇਣੀ ਪੈਂਦੀ ਹੈ।

- Azure OpenAI Service ਸਰੋਤ ਬਣਾਓ। [ਇਹ ਗਾਈਡ](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ਵੇਖੋ ਕਿ ਸਰੋਤ ਕਿਵੇਂ ਬਣਾਉਣਾ ਹੈ।

### API ਕੁੰਜੀ ਅਤੇ ਐਂਡਪੌਇੰਟ ਲੱਭੋ

ਹੁਣ ਤੁਹਾਨੂੰ ਆਪਣੀ `openai` ਲਾਇਬ੍ਰੇਰੀ ਨੂੰ ਦੱਸਣਾ ਪਵੇਗਾ ਕਿ ਕਿਹੜੀ API ਕੁੰਜੀ ਵਰਤਣੀ ਹੈ। ਆਪਣੀ API ਕੁੰਜੀ ਲੱਭਣ ਲਈ, Azure OpenAI ਸਰੋਤ ਦੇ "Keys and Endpoint" ਭਾਗ 'ਚ ਜਾ ਕੇ "Key 1" ਦੀ ਵੈਲਿਊ ਕਾਪੀ ਕਰੋ।

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ਹੁਣ ਜਦ ਤੁਹਾਡੇ ਕੋਲ ਇਹ ਜਾਣਕਾਰੀ ਹੈ, ਆਓ ਲਾਇਬ੍ਰੇਰੀਆਂ ਨੂੰ ਇਹ ਵਰਤਣ ਲਈ ਦੱਸਦੇ ਹਾਂ।

> [!NOTE]
> ਆਪਣੀ API ਕੁੰਜੀ ਨੂੰ ਆਪਣੇ ਕੋਡ ਤੋਂ ਵੱਖ ਰੱਖਣਾ ਚੰਗਾ ਹੈ। ਤੁਸੀਂ ਇਹ ਵਾਤਾਵਰਣ ਚਲਣ ਵਾਲੀਆਂ ਵੈਰੀਏਬਲਾਂ ਰਾਹੀਂ ਕਰ ਸਕਦੇ ਹੋ।
> - .env ਫਾਈਲ ਵਿੱਚ `AZURE_OPENAI_API_KEY` ਵਾਤਾਵਰਣ ਵੈਰੀਏਬਲ ਨੂੰ ਆਪਣੀ API ਕੁੰਜੀ ਤੇ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਇਸ ਕੋਰਸ ਦੇ ਪਿਛਲੇ ਅਭਿਆਸ ਕਰ ਲਏ ਹਨ, ਤਾਂ ਤੁਸੀਂ ਤਿਆਰ ਹੋ।

### Azure ਦੀ ਸੰਰਚਨਾ ਸੈੱਟ ਕਰੋ

ਜੇ ਤੁਸੀਂ Azure OpenAI ਵਰਤ ਰਹੇ ਹੋ, ਤਾਂ ਇੰਝ ਸੰਰਚਨਾ ਕਰੋ:

```python
client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  azure_endpoint = os.environ('AZURE_OPENAI_ENDPOINT')
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ਉੱਪਰ ਅਸੀਂ ਇਹ ਸੈੱਟ ਕਰ ਰਹੇ ਹਾਂ:

- `api_key`, ਇਹ ਤੁਹਾਡੀ API ਕੁੰਜੀ ਹੈ ਜੋ Azure Portal ਵਿੱਚ ਮਿਲਦੀ ਹੈ।
- `api_version`, ਇਹ API ਦਾ ਵਰਜਨ ਹੈ ਜੋ ਤੁਸੀਂ ਵਰਤਣਾ ਚਾਹੁੰਦੇ ਹੋ। ਲਿਖਣ ਸਮੇਂ, ਨਵਾਂ ਵਰਜਨ `2023-10-01-preview` ਹੈ।

## ਕਸਰਤ - ਇੱਕ ਰੈਸੀਪੀ ਜਨਰੇਟਰ ਬਣਾਓ

1. ਮੌਜੂਦਾ *app.py* ਫਾਈਲ ਨੂੰ ਸ਼ੁਰੂਆਤ ਵਜੋਂ ਵਰਤੋ
1. `prompt` ਵੇਰੀਏਬਲ ਲੱਭੋ ਅਤੇ ਉਸਦੇ ਕੋਡ ਨੂੰ ਹੇਠਾਂ ਦਿੱਤੇ ਤਰੀਕੇ ਨਾਲ ਬਦਲੋ:

ਵਧੀਆ, ਆਓ ਵੇਖੀਏ ਕਿ ਅਸੀਂ ਚੀਜ਼ਾਂ ਨੂੰ ਕਿਵੇਂ ਬਿਹਤਰ ਕਰ ਸਕਦੇ ਹਾਂ। ਚੀਜ਼ਾਂ ਨੂੰ ਬਿਹਤਰ ਬਣਾਉਣ ਲਈ, ਅਸੀਂ ਚਾਹੁੰਦੇ ਹਾਂ ਕਿ ਕੋਡ ਲਚਕੀਲਾ ਹੋਵੇ, ਤਾਂ ਜੋ ਸਮੱਗਰੀਆਂ ਅਤੇ ਰੈਸੀਪੀਆਂ ਦੀ ਗਿਣਤੀ ਨੂੰ ਆਸਾਨੀ ਨਾਲ ਬਦਲਿਆ ਜਾਂ ਵਧਾਇਆ ਜਾ ਸਕੇ।


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
