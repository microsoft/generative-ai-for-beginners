# ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਸ ਬਣਾਓ

ਤੁਸੀਂ ਇਸ ਕਰੀਕੁਲਮ ਰਾਹੀਂ ਹੁਣ ਤੱਕ ਦੇਖਿਆ ਹੈ ਕਿ ਮੂਲ ਭਾਵਨਾਵਾਂ ਹਨ ਜਿਵੇਂ ਕਿ ਪ੍ਰਾਂਪਟ ਅਤੇ ਇੱਕ ਪੂਰਾ ਵਿਸ਼ਾ "ਪ੍ਰਾਂਪਟ ਇੰਜੀਨੀਅਰਿੰਗ" ਵੀ ਹੈ। ਬਹੁਤ ਸਾਰੇ ਟੂਲ ਜਿਵੇਂ ਕਿ ChatGPT, Office 365, Microsoft Power Platform ਅਤੇ ਹੋਰ, ਤੁਹਾਡੀ ਸਹਾਇਤਾ ਕਰਦੇ ਹਨ ਪ੍ਰਾਂਪਟਾਂ ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਕੁਝ ਪ੍ਰਾਪਤ ਕਰਨ ਵਿੱਚ।

ਕੋਈ ਐਪ ਵਿੱਚ ਇਹ ਤਜ਼ਰਬਾ ਸ਼ਾਮਲ ਕਰਨ ਲਈ, ਤੁਹਾਨੂੰ ਪ੍ਰਾਂਪਟ, ਕੰਪਲੀਸ਼ਨ ਸਮਝਣਾ ਲਾਜ਼ਮੀ ਹੈ ਅਤੇ ਇੱਕ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਚੋਣ ਕਰਨੀ ਪੈਂਦੀ ਹੈ। ਇਨ੍ਹਾਂ ਸਭ ਚੀਜਾਂ ਦਾ ਇਹੈਠਿਕ ਅਧਿਆਇ ਵਿੱਚ ਸਿੱਖਿਅਾ ਜਾਵੇਗਾ।

## ਪਰਿਚਯ

ਇਸ ਅਧਿਆਇ ਵਿੱਚ, ਤੁਸੀਂ:

- openai ਲਾਇਬ੍ਰੇਰੀ ਅਤੇ ਇਸਦੇ ਮੂਲ ਭਾਵਨਾਵਾਂ ਬਾਰੇ ਸਿੱਖੋਗੇ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇੱਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓਗੇ।
- ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਉਣ ਲਈ ਪ੍ਰਾਂਪਟ, ਟੈਂਪਰੈਚਰ ਅਤੇ ਟੋਕਨ ਵਰਗੀਆਂ ਭਾਵਨਾਵਾਂ ਦੀ ਵਰਤੋਂ ਕਿਵੇਂ ਕਰਨੀ ਹੈ, ਸਮਝੋਗੇ।

## ਸਿੱਖਣ ਦੇ ਲਕੜੀ

ਇਸ ਪਾਠ ਦੇ ਅੰਤ ਵਿੱਚ, ਤੁਸੀਂ ਸਮਰੱਥ ਹੋਵੋਗੇ:

- ਸਮਝਾਓ ਕਿ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੁੰਦੀ ਹੈ।
- openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਇਕ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਬਣਾਓ।
- ਆਪਣੇ ਐਪ ਨੂੰ ਹੋਰ ਜ਼ਿਆਦਾ ਜਾਂ ਘੱਟ ਟੋਕਨ ਵਰਤਣ ਲਈ ਸੰਰਚਿਤ ਕਰੋ ਅਤੇ ਆਉਟਪੁੱਟ ਵਿੱਚ ਬਦਲਾਅ ਲਈ ਟੈਂਪਰੈਚਰ ਵੀ ਬਦਲੋ।

## ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਕੀ ਹੈ?

ਆਮ ਤੌਰ 'ਤੇ ਜਦੋਂ ਤੁਸੀਂ ਇੱਕ ਐਪ ਬਣਾਉਂਦੇ ਹੋ ਤਾਂ ਇਸ ਵਿੱਚ ਕੋਈ ਨਾ ਕੋਈ ਇੰਟਰਫੇਸ ਹੁੰਦਾ ਹੈ, ਜਿਵੇਂ ਕਿ ਹੇਠਾਂ ਦਿੱਤਾ ਗਿਆ ਹੈ:

- ਕਮਾਂਡ-ਅਧਾਰਿਤ। ਕਨਸੋਲ ਐਪਸ ਉਹ ਐਪ ਹੁੰਦੇ ਹਨ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ ਅਤੇ ਉਹ ਕੰਮ ਕਰਾ ਦਿੰਦੇ ਹਨ। ਉਦਾਹਰਨ ਵਜੋਂ, `git` ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਹੈ।
- ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (UI)। ਕੁਝ ਐਪਸ ਵਿੱਚ ਗ੍ਰਾਫਿਕਲ ਯੂਜ਼ਰ ਇੰਟਰਫੇਸ (GUI) ਹੁੰਦਾ ਹੈ ਜਿੱਥੇ ਤੁਸੀਂ ਬਟਨ ਕਲਿੱਕ ਕਰਦੇ ਹੋ, ਟੈਕਸਟ ਦਰਜ ਕਰਦੇ ਹੋ, ਵਿਕਲਪ ਚੁਣਦੇ ਹੋ ਅਤੇ ਹੋਰ।

### ਕਨਸੋਲ ਅਤੇ UI ਐਪਸ ਸੀਮਿਤ ਹੁੰਦੇ ਹਨ

ਇਸ ਦੀ ਤੁਲਨਾ ਕਰੋ ਇੱਕ ਕਮਾਂਡ-ਅਧਾਰਿਤ ਐਪ ਨਾਲ ਜਿੱਥੇ ਤੁਸੀਂ ਕਮਾਂਡ ਟਾਈਪ ਕਰਦੇ ਹੋ:

- **ਇਹ ਸੀਮਿਤ ਹੈ**। ਤੁਸੀਂ ਕੋਈ ਵੀ ਕਮਾਂਡ ਨਹੀਂ ਟਾਈਪ ਕਰ ਸਕਦੇ, ਬਸ ਉਹ ਹੀ ਜਿਹੜੀਆਂ ਐਪ ਸਹਾਇਤਾ ਕਰਦਾ ਹੈ।
- **ਭਾਸ਼ਾ-ਵਿਸ਼ੇਸ਼**। ਕੁਝ ਐਪ ਬਹੁਤ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦੇ ਹਨ, ਪਰ ਡੀਫੌਲਟ ਰੂਪ ਵਿੱਚ ਐਪ ਇੱਕ ਖਾਸ ਭਾਸ਼ਾ ਲਈ ਬਣਾਇਆ ਗਿਆ ਹੁੰਦਾ ਹੈ, ਭਾਵੇਂ ਤੁਸੀਂ ਹੋਰ ਭਾਸ਼ਾਵਾਂ ਦਾ ਸਹਾਰਾ ਦੇ ਸਕਦੇ ਹੋ।

### ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪਸ ਦੇ ਲਾਭ

ਤਾਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵੱਖਰਾ ਕਿਵੇਂ ਹੈ?

ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਵਿੱਚ ਤੁਹਾਡੇ ਕੋਲ ਵੱਧ ਲਚਕੀਲਾ ਹੁੰਦਾ ਹੈ, ਤੁਸੀਂ ਕਮਾਂਡਾਂ ਜਾਂ ਖਾਸ ਇੰਪੁੱਟ ਭਾਸ਼ਾ ਨਾਲ ਸੀਮਿਤ ਨਹੀਂ ਹੋ, ਪਰ ਕੁਦਰਤੀ ਭਾਸ਼ਾ ਵਰਤ ਕੇ ਐਪ ਨਾਲ ਇੰਟਰੈਕਟ ਕਰ ਸਕਦੇ ਹੋ। ਇੱਕ ਹੋਰ ਫਾਇਦਾ ਇਹ ਹੈ ਕਿ ਤੁਸੀਂ ਪਹਿਲਾਂ ਤੋਂ ਹੀ ਇੱਕ ਡਾਟਾ ਸਰੋਤ ਨਾਲ ਸੰਪਰਕ ਵਿੱਚ ਹੋ ਜੋ ਵੱਡੇ ਪਾਠ ਸਮੱਗਰੀ 'ਤੇ ਟ੍ਰੇਨ ਕੀਤਾ ਗਿਆ ਹੈ, ਜਦਕਿ ਆਮ ਐਪ ਡੇਟਾਬੇਸ ਵਿੱਚ ਮੌਜੂਦ ਹਦਾਇਤਾਂ ਤੱਕ ਸੀਮਿਤ ਹੋ ਸਕਦਾ ਹੈ।

### ਮੈਂ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਐਪ ਨਾਲ ਕੀ ਬਣਾ ਸਕਦਾ ਹਾਂ?

ਤੁਸੀਂ ਬਹੁਤ ਕੁਝ ਬਣਾ ਸਕਦੇ ਹੋ। ਉਦਾਹਰਨ ਵਜੋਂ:

- **ਚੈਟਬੋਟ**। ਇੱਕ ਚੈਟਬੋਟ ਜੋ ਤੁਹਾਡੇ ਕੰਪਨੀ ਅਤੇ ਉਨ੍ਹਾਂ ਦੇ ਉਤਪਾਦਾਂ ਬਾਰੇ ਪ੍ਰਸ਼ਨਾਂ ਦੇ ਜਵਾਬ ਦਿੰਦਾ ਹੋਵੇ, ਵਧੀਆ ਸਹਾਇਕ ਹੋ ਸਕਦਾ ਹੈ।
- **ਸਹਾਇਕ**। LLM ਟੈਕਸਟ ਦਾ ਸਾਰ ਵੇਖਾਉਣ, ਟੈਕਸਟ ਵਿੱਚੋਂ ਜ਼ਰੂਰੀ ਜਾਣਕਾਰੀਆਂ ਪ੍ਰਾਪਤ ਕਰਨ, ਰੇਜ਼ੂਮੇ ਵਰਗਾ ਟੈਕਸਟ ਤਿਆਰ ਕਰਨ ਜਾਂ ਹੋਰ ਚੀਜ਼ਾਂ ਵਿੱਚ ਬਹੁਤ ਵਧੀਆ ਹਨ।
- **ਕੋਡ ਸਹਾਇਕ**। ਤੁਸੀਂ ਜਿਹੜਾ ਭਾਸ਼ਾ ਮਾਡਲ ਵਰਤੋਂਗੇ, ਉਸ ਦੇ ਅਨੁਸਾਰ ਇੱਕ ਕੋਡ ਸਹਾਇਕ ਤਿਆਰ ਕਰ ਸਕਦੇ ਹੋ ਜੋ ਤੁਸੀਂ ਕੋਡ ਲਿਖਣ ਵਿੱਚ ਮਦਦ ਕਰੇ। ਉਦਾਹਰਨ ਵਜੋਂ, ਤੁਸੀਂ GitHub Copilot ਜਾਂ ChatGPT ਵਰਗੇ ਉਤਪਾਦਾਂ ਦੀ ਵਰਤੋਂ ਕਰ ਸਕਦੇ ਹੋ।

## ਮੈਂ ਕਿਵੇਂ ਸ਼ੁਰੂ ਕਰ ਸਕਦਾ ਹਾਂ?

ਠੀਕ ਹੈ, ਤੁਹਾਨੂੰ ਕਿਸੇ LLM ਨਾਲ ਇੰਟੀਗ੍ਰੇਟ ਕਰਨ ਦਾ ਤਰੀਕਾ ਲੱਭਣਾ ਪਵੇਗਾ ਜੋ ਆਮ ਤੌਰ 'ਤੇ ਦੋ ਤਰੀਕਿਆਂ ਵਿੱਚੋਂ ਇੱਕ ਹੁੰਦਾ ਹੈ:

- ਏਪੀਆਈ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਇੱਥੇ ਤੁਸੀਂ ਆਪਣਾ ਪ੍ਰਾਂਪਟ ਲੈ ਕੇ ਵੈੱਬ ਰਿਕਵੈਸਟ ਬਣਾਉਂਦੇ ਹੋ ਅਤੇ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਪ੍ਰਾਪਤ ਕਰਦੇ ਹੋ।
- ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਵਰਤੋਂ ਕਰੋ। ਲਾਇਬ੍ਰੇਰੀ API ਕਾਲਾਂ ਨੂੰ ਇੰਕੈਪਸੂਲੇਟ ਕਰਕੇ ਉਹਨਾਂ ਨੂੰ ਆਸਾਨ ਬਣਾਉਂਦੀ ਹੈ।

## ਲਾਇਬ੍ਰੇਰੀਜ਼/SDKs

LLM ਨਾਲ ਕੰਮ ਕਰਨ ਲਈ ਕੁਝ ਪ੍ਰਸਿੱਧ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ ਕਿ:

- **openai**, ਇਸ ਲਾਇਬ੍ਰੇਰੀ ਨਾਲ ਆਪਣੇ ਮਾਡਲ ਨਾਲ ਜੁੜਨਾ ਅਤੇ ਪ੍ਰਾਂਪਟ ਭੇਜਣਾ ਆਸਾਨ ਹੋ ਜਾਂਦਾ ਹੈ।

ਇੱਥੇ ਵੱਡੇ ਪੱਧਰ 'ਤੇ ਕੰਮ ਕਰਨ ਵਾਲੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਹਨ ਜਿਵੇਂ:

- **Langchain**। Langchain ਬਹੁਤ ਮਸ਼ਹੂਰ ਹੈ ਅਤੇ ਪਾਇਥਨ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦੀ ਹੈ।
- **Semantic Kernel**। Semantic Kernel ਮਾਈਕ੍ਰੋਸਾਫਟ ਦੀ ਲਾਇਬ੍ਰੇਰੀ ਹੈ ਜੋ C#, ਪਾਇਥਨ ਅਤੇ ਜਾਵਾ ਭਾਸ਼ਾਵਾਂ ਨੂੰ ਸਹਾਇਤਾ ਦਿੰਦੀ ਹੈ।

## openai ਦੀ ਵਰਤੋਂ ਕਰਕੇ ਪਹਿਲੀ ਐਪ

ਆਓ ਵੇਖਦੇ ਹਾਂ ਕਿ ਅਸੀਂ ਆਪਣੀ ਪਹਿਲੀ ਐਪ ਕਿਵੇਂ ਬਣਾਈਏ, ਕਿਹੜੀਆਂ ਲਾਇਬ੍ਰੇਰੀਆਂ ਚਾਹੀਦੀਆਂ ਹਨ, ਕਿੰਨੀ ਲੋੜ ਹੈ ਅਤੇ ਹੋਰ।

### openai ਇੰਸਟਾਲ ਕਰੋ

  > [!NOTE] ਇਹ ਕਦਮ ਜਰੂਰੀ ਨਹੀ ਹੈ ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Devcontainer ਵਿੱਚ ਚਲਾ ਰਹੇ ਹੋ


OpenAI ਜਾਂ Azure OpenAI ਨਾਲ ਸੰਪਰਕ ਕਰਨ ਲਈ ਬਹੁਤ ਲਾਇਬ੍ਰੇਰੀਆਂ ਉਪਲਬਧ ਹਨ। ਇਹ ਵੀ ਸੰਭਵ ਹੈ ਕਿ ਕਈ ਪ੍ਰੋਗਰਾਮਿੰਗ ਭਾਸ਼ਾਵਾਂ ਵਰਤੀ ਜਾ ਸਕਦੀਆਂ ਹਨ ਜਿਵੇਂ ਕਿ C#, Python, JavaScript, Java ਆਦਿ।
ਅਸੀਂ `openai` ਪਾਇਥਨ ਲਾਇਬ੍ਰੇਰੀ ਦੀ ਚੋਣ ਕੀਤੀ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `pip` ਨਾਲ ਇਸਨੂੰ ਇੰਸਟਾਲ ਕਰਾਂਗੇ।

```bash
pip install openai
```

ਜੇ ਤੁਸੀਂ ਇਹ ਨੋਟਬੁੱਕ Codespaces ਜਾਂ Dev Container ਵਿੱਚ ਨਹੀਂ ਚਲਾ ਰਹੇ, ਤਾਂ ਤੁਹਾਨੂੰ [Python](https://www.python.org/) ਆਪਣੇ ਮਸ਼ੀਨ 'ਤੇ ਇੰਸਟਾਲ ਕਰਨਾ ਪਵੇਗਾ।

### ਸਰੋਤ ਬਣਾਓ

ਜੇ ਤੁਹਾਡੇ ਕੋਲ ਪਹਿਲਾਂ ਨਹੀਂ ਹੈ, ਤਾਂ ਇਹ ਹੇਠਾਂ ਦਿੱਤੇ ਕਦਮ ਅਮਲ ਕਰੋ:

- Azure 'ਤੇ ਇੱਕ ਅਕਾਉਂਟ ਬਣਾਓ <https://azure.microsoft.com/free/>।
- Microsoft Foundry ਵਿੱਚ ਇੱਕ Azure OpenAI ਸਰੋਤ ਬਣਾਓ [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst)। ਇਸ ਮਾਰਗਦਰਸ਼ਿਕਾ ਨੂੰ ਦੇਖੋ ਕਿ ਕਿਵੇਂ [ਸਰੋਤ ਬਣਾਇਆ ਜਾਵੇ](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst)।


### API ਕੀ ਅਤੇ ਐਂਡਪੌਇੰਟ ਲੱਭੋ

ਇਸ ਸਮੇਂ, ਤੁਹਾਨੂੰ ਆਪਣੀ `openai` ਲਾਇਬ੍ਰੇਰੀ ਦੱਸਣੀ ਪਵੇਗੀ ਕਿ ਕਿਹੜੀ API ਕੀ ਵਰਤਣੀ ਹੈ। ਆਪਣੀ API ਕੀ ਲੱਭਣ ਲਈ, ਆਪਣੇ Azure OpenAI ਸਰੋਤ ਦੇ "Keys and Endpoint" ਸੈਕਸ਼ਨ ਵਿੱਚ ਜਾਓ ਅਤੇ "Key 1" ਦੀ ਕਾਪੀ ਕਰੋ।

  ![Azure Portal ਵਿੱਚ Keys and Endpoint ਸਰੋਤ ਬਲੇਡ](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ਹੁਣ ਜਦ ਤੁਹਾਡੇ ਕੋਲ ਇਸ ਜਾਣਕਾਰੀ ਦੀ ਕਾਪੀ ਹੈ, ਤਾਂ ਲਾਇਬ੍ਰੇਰੀਜ਼ ਨੂੰ ਇਹ ਵਰਤਣ ਲਈ ਦਿਓ।

> [!NOTE]
> ਇਹ ਵਧੀਆ ਹੈ ਕਿ ਤੁਸੀਂ ਆਪਣੀ API ਕੀ ਨੂੰ ਆਪਣੇ ਕੋਡ ਤੋਂ ਵੱਖ ਕਰ ਦਿਓ। ਤੁਸੀਂ ਇਸਦਾ ਉਪਯੋਗ ਇੰਵਾਇਰਨਮੈਂਟ ਵੈਰੀਏਬਲ ਨਾਲ ਕਰ ਸਕਦੇ ਹੋ।
> - ਆਪਣੀ API ਕੀ ਲਈ ਇੰਵਾਇਰਨਮੈਂਟ ਵੈਰੀਏਬਲ `AZURE_OPENAI_API_KEY` ਨੂੰ ਆਪਣੇ .env ਫਾਈਲ ਵਿੱਚ ਸੈੱਟ ਕਰੋ। ਜੇ ਤੁਸੀਂ ਇਸ ਕੋਰਸ ਦੇ ਪਹਿਲੇ ਅਭਿਆਸ ਮੁਕੰਮਲ ਕਰ ਚੁੱਕੇ ਹੋ, ਤਾਂ ਤੁਸੀਂ ਤਿਆਰ ਹੋ।


### Azure ਸੰਰਚਨਾ ਸੈੱਟਅੱਪ ਕਰੋ

ਜੇ ਤੁਸੀਂ Azure OpenAI ਵਰਤ ਰਹੇ ਹੋ, ਤਾਂ ਇੱਥੇ ਸੰਰਚਨਾ ਸੈੱਟਅੱਪ ਹੈ। Responses API Azure OpenAI **v1 endpoint** ਤੋਂ ਸੇਵਾ ਪ੍ਰਦਾਨ ਕਰਦਾ ਹੈ, ਇਸ ਲਈ ਅਸੀਂ `OpenAI` ਕਲਾਇੰਟ ਨੂੰ `<your-endpoint>/openai/v1/` 'ਤੇ ਨਿਸ਼ਾਨਾ ਕਰਦੇ ਹਾਂ:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ਹੇਠਾਂ ਅਸੀਂ ਇਹ ਡਿਫਾਈਨ ਕਰ ਰਹੇ ਹਾਂ:

- `api_key`, ਇਹ ਤੁਹਾਡੀ API ਕੀ ਹੈ ਜੋ Azure Portal ਵਿੱਚ ਮਿਲੀ ਹੈ।
- `base_url`, ਇਹ ਤੁਹਾਡਾ Azure OpenAI ਐਂਡਪੌਇੰਟ ਹੈ ਜਿਸ ਨਾਲ `/openai/v1/` ਜੋੜਿਆ ਗਿਆ ਹੈ। ਤੁਸੀਂ ਆਪਣਾ ਐਂਡਪੌਇੰਟ Azure Portal ਦੇ ਨਾਲ API ਕੀ ਦੇਖ ਕੇ ਲੱਭ ਸਕਦੇ ਹੋ। v1 ਐਂਡਪੌਇੰਟ ਵਰਤਣ ਨਾਲ ਤੁਹਾਨੂੰ `api_version` ਪਾਸ ਕਰਨ ਦੀ ਲੋੜ ਨਹੀਂ ਰਹਿੰਦੀ।
- `deployment`, ਇਹ ਤੁਹਾਡੇ ਦੁਆਰਾ Foundry ਪੋਰਟਲ ਵਿੱਚ ਬਣਾਏ ਮਾਡਲ ਡਿਪਲੋਇਮੈਂਟ ਦਾ ਨਾਮ ਹੈ।

> [!NOTE]
> `os.environ` ਇੱਕ ਮੈਪ ਹੈ ਜੋ ਇੰਵਾਇਰਨਮੈਂਟ ਵੈਰੀਏਬਲ ਪੜ੍ਹਦਾ ਹੈ। ਤੁਸੀਂ ਇਸ ਨੂੰ ਵਰਤ ਕੇ ਇੰਵਾਇਰਨਮੈਂਟ ਵੈਰੀਏਬਲ ਜਿਵੇਂ ਕਿ `AZURE_OPENAI_API_KEY` ਅਤੇ `AZURE_OPENAI_ENDPOINT` ਪੜ੍ਹ ਸਕਦੇ ਹੋ।

## ਟੈਕਸਟ ਜਨਰੇਟ ਕਰੋ

ਟੈਕਸਟ ਬਣਾਉਣ ਦਾ ਤਰੀਕਾ `responses.create` ਮੈਥਡ ਦੀ ਵਰਤੋਂ ਕਰਨਾ ਹੈ। ਹੇਠਾਂ ਇੱਕ ਉਦਾਹਰਨ ਹੈ:

```python
prompt = "Complete the following: Once upon a time there was a"

print(response.output_text)
```

ਉਪਰੋਕਤ ਕੋਡ ਵਿੱਚ, ਅਸੀਂ ਇੱਕ ਰਿਸਪਾਂਸ ਆਬਜੈਕਟ ਬਣਾਉਂਦੇ ਹਾਂ ਅਤੇ ਮਾਡਲ ਅਤੇ ਪ੍ਰਾਂਪਟ ਦਿੰਦੇ ਹਾਂ। ਫਿਰ ਅਸੀਂ ਜਨਰੇਟ ਕੀਤਾ ਟੈਕਸਟ ਪ੍ਰਿੰਟ ਕਰਦੇ ਹਾਂ।

### ਚੈਟ ਜਵਾਬ

Responses API ਦੋਹਾਂ ਸਿੰਗਲ-ਟਰਨ ਟੈਕਸਟ ਜਨਰੇਸ਼ਨ ਅਤੇ ਮਲਟੀ-ਟਰਨ ਚੈਟਬੋਟਸ ਲਈ ਵਧੀਆ ਹੈ - ਤੁਸੀਂ ਵਰਤੋਂਕਾਰ ਦੇ ਵੱਖ-ਵੱਖ ਸੁਨੇਹੇ `input` ਲਿਸਟ ਵਿੱਚ ਦੇ ਸਕਦੇ ਹੋ ਤਾਂ ਜੋ ਇੱਕ ਗੱਲਬਾਤ ਬਣਾਈ ਜਾ ਸਕੇ:

