# Komut Mühendisliğine Giriş
Komut mühendisliği, doğal dil işleme görevleri için komutların tasarımı ve optimize edilmesi sürecidir. Doğru komutları seçmeyi, parametrelerini ayarlamayı ve performanslarını değerlendirmeyi içerir. Komut mühendisliği, NLP modellerinde yüksek doğruluk ve verimlilik elde etmek için çok önemlidir. Bu bölümde, keşif için OpenAI modelleri kullanarak komut mühendisliğinin temellerini inceleyeceğiz.


### Alıştırma 1: Tokenizasyon
OpenAI'nin açık kaynak hızlı tokenize edicisi tiktoken ile Tokenizasyonu keşfedin
Daha fazla örnek için [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) adresine bakın.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### Alıştırma 2: OpenAI API Anahtarı Kurulumunu Doğrulama

Aşağıdaki kodu çalıştırarak OpenAI uç noktanızın doğru şekilde ayarlandığını doğrulayın. Kod, basit bir temel istemi dener ve tamamlamayı doğrular. `oh say can you see` girdisi `by the dawn's early light..` gibi bir tamamlamayla sonuçlanmalıdır.


````python
# Uses the OpenAI client against the Azure OpenAI (Microsoft Foundry) v1 endpoint
# with the Responses API. See https://aka.ms/openai/start

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### Alıştırma 4: Talimat Tabanlı 
Birincil içeriği ayarlamak için "text" değişkenini kullanın 
ve bu birincil içerikle ilgili bir talimat sağlamak için "prompt" değişkenini kullanın.

Burada modelden metni ikinci sınıf öğrencisi için özetlemesini istiyoruz


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### Alıştırma 5: Karmaşık İstek  
Sistem, kullanıcı ve asistan mesajları içeren bir istek deneyin  
Sistem, asistanın bağlamını ayarlar  
Kullanıcı ve Asistan mesajları çok turlu konuşma bağlamı sağlar  

Asistan kişiliğinin sistem bağlamında "alaycı" olarak ayarlandığına dikkat edin.  
Farklı bir kişilik bağlamı deneyin. Veya farklı bir giriş/çıkış mesaj dizisi deneyin  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
