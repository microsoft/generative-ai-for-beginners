## Metin oluşturma uygulaması nedir?

Normalde bir uygulama oluşturduğunuzda aşağıdaki gibi bir arayüze sahiptir:

- Komut tabanlı. Konsol uygulamaları, bir komut yazdığınız ve bir görevi gerçekleştirdiği tipik uygulamalardır. Örneğin, `git` komut tabanlı bir uygulamadır.
- Kullanıcı arayüzü (UI). Bazı uygulamalar, düğmelere tıklayabileceğiniz, metin girebileceğiniz, seçenekleri seçebileceğiniz grafiksel kullanıcı arayüzlerine (GUI) sahiptir.

### Konsol ve UI uygulamaları sınırlıdır

Bir komut tabanlı uygulama ile karşılaştırın:

- **Sınırlıdırlar**. Herhangi bir komutu yazamazsınız, yalnızca uygulamanın desteklediği komutları yazabilirsiniz.
- **Dil spesifik**. Bazı uygulamalar birçok dili destekler, ancak varsayılan olarak uygulama belirli bir dil için oluşturulmuştur, daha fazla dil desteği ekleseniz bile.

### Metin oluşturma uygulamalarının avantajları

Peki metin oluşturma uygulaması nasıl farklıdır?

Metin oluşturma uygulamasında daha esneksiniz, belirli komutlara veya belirli bir giriş diline bağlı kalmazsınız. Bunun yerine uygulamayla doğal dili kullanarak etkileşim kurabilirsiniz. Bir diğer avantaj ise zaten çok büyük bir bilgi derlemesi üzerinde eğitilmiş bir veri kaynağıyla etkileşimde bulunuyor olmanızdır, oysa geleneksel bir uygulama veritabanındaki bilgilerle sınırlı olabilir.

### Metin oluşturma uygulaması ile neler yapabilirim?

Birçok şey oluşturabilirsiniz. Örneğin:

- **Bir sohbet botu**. Şirketiniz ve ürünleri hakkında soruları yanıtlayan bir sohbet botu iyi bir eşleşme olabilir.
- **Yardımcı**. LLM'ler metinleri özetlemek, metinden çıkarımlar yapmak, özgeçmiş gibi metinler üretmek gibi şeylerde harikadır.
- **Kod asistanı**. Kullandığınız dil modeline bağlı olarak, yazmaya yardımcı olan bir kod asistanı oluşturabilirsiniz. Örneğin, GitHub Copilot veya ChatGPT gibi ürünleri kod yazmaya yardımı için kullanabilirsiniz.

### Azure yapılandırmasını ayarlama

Azure OpenAI kullanıyorsanız, yapılandırmayı şöyle yaparsınız. Yanıtlar API'sı Azure OpenAI **v1 uç noktasından** sağlanır, bu nedenle `OpenAI` istemcisini `<your-endpoint>/openai/v1/` adresine işaret ederiz:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Yukarıda şu ayarları yapıyoruz:

- `api_key`, bu Azure Portalda bulduğunuz API anahtarınızdır.
- `base_url`, bu Azure OpenAI uç noktanızdır, `/openai/v1/` eklenmiş olarak. Uç noktayı Azure Portalda API anahtarınızın yanında bulabilirsiniz. v1 uç noktasını kullanmak `api_version` göndermeniz gerekmediği anlamına gelir.
- `deployment`, Foundry portalında oluşturduğunuz model dağıtımının adıdır.

> [!NOTE]
> `os.environ` ortam değişkenlerini okuyan bir haritadır. `AZURE_OPENAI_API_KEY` ve `AZURE_OPENAI_ENDPOINT` gibi ortam değişkenlerini okumak için kullanabilirsiniz.

## Metin oluşturun

Metin oluşturmanın yolu `responses.create` yöntemini kullanmaktır. İşte bir örnek:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yukarıdaki kodda bir yanıt nesnesi oluştururuz ve kullanmak istediğimiz modeli ve istemi geçiririz. Sonra oluşturulan metni yazdırırız.

### Sohbet yanıtları

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

We now have a working app capable of producing recipes and it's flexible as it relies on inputs from the user, both on the number of recipes but also the ingredients used.

To further improve it, we want to add the following:

- **Filter out ingredients**. We want to be able to filter out ingredients we don't like or are allergic to. To accomplish this change, we can edit our existing prompt and add a filter condition to the end of it like so:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Above, we add `{filter}` to the end of the prompt and we also capture the filter value from the user.
