## Metin oluşturma uygulaması nedir?

Normalde bir uygulama oluşturduğunuzda, aşağıdakilere benzer bir arayüzü olur:

- Komut tabanlı. Konsol uygulamaları, bir komut yazıp görevi gerçekleştiren tipik uygulamalardır. Örneğin, `git` komut tabanlı bir uygulamadır.
- Kullanıcı arayüzü (UI). Bazı uygulamalar, düğmelere tıklayabileceğiniz, metin girebileceğiniz, seçenekleri seçebileceğiniz ve daha fazlasını yapabileceğiniz grafik kullanıcı arayüzlerine (GUI) sahiptir.

### Konsol ve UI uygulamaları sınırlıdır

Bir komut tabanlı uygulamayla karşılaştırın, bir komut yazarsınız:

- **Sınırlıdır**. İstediğiniz herhangi bir komutu yazamazsınız, yalnızca uygulamanın desteklediği komutları yazabilirsiniz.
- **Dil spesifik**. Bazı uygulamalar birçok dili destekler, ancak varsayılan olarak uygulama belirli bir dil için oluşturulmuştur, hatta daha fazla dil desteği ekleyebilmenize rağmen.

### Metin oluşturma uygulamalarının faydaları

Peki metin oluşturma uygulaması nasıl farklıdır?

Metin oluşturma uygulamasında daha fazla esnekliğiniz vardır, belirli komut setleri veya belirli bir giriş dili ile sınırlı değilsiniz. Bunun yerine, uygulama ile etkileşim için doğal dili kullanabilirsiniz. Bir diğer avantaj da, zaten çok geniş bir bilgi korpusunda eğitilmiş bir veri kaynağıyla etkileşimde olmanızdır, oysa geleneksel bir uygulama bir veritabanındakiyle sınırlı olabilir.

### Metin oluşturma uygulamasıyla ne tür şeyler yapabilirim?

Birçok şey inşa edebilirsiniz. Örneğin:

- **Bir sohbet botu**. Şirketiniz ve ürünleri hakkında soruları yanıtlayan bir sohbet botu iyi bir eşleşme olabilir.
- **Yardımcı**. Büyük dil modelleri (LLM'ler), metin özetleme, metinden içgörü elde etme, özgeçmiş gibi metinler üretme gibi görevlerde harikadır.
- **Kod asistanı**. Kullandığınız dil modeline bağlı olarak, size kod yazmada yardımcı olan bir kod asistanı oluşturabilirsiniz. Örneğin, GitHub Copilot ve ChatGPT gibi ürünleri kullanarak kod yazmanıza yardımcı olabilir.

### Azure yapılandırmasını ayarlama

Azure OpenAI kullanıyorsanız, yapılandırmayı şöyle ayarlarsınız. Yanıtlar API'si Azure OpenAI **v1 uç noktasından** sağlanır, bu nedenle `OpenAI` istemcisini `<your-endpoint>/openai/v1/` adresine yönlendiriyoruz:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Yukarıda şunları ayarlıyoruz:

- `api_key`, bu Azure Portal’da bulunan API anahtarınızdır.
- `base_url`, bu Azure OpenAI uç noktanızdır, sonuna `/openai/v1/` eklenmiştir. Uç noktayı Azure Portal’da API anahtarınızın yanında bulabilirsiniz. v1 uç noktayı kullanmak, `api_version` geçirmenize artık gerek kalmadığı anlamına gelir.
- `deployment`, bu Foundry portalında oluşturduğunuz model dağıtımının adıdır.

> [!NOTE]
> `os.environ`, ortam değişkenlerini okuyan bir haritalamadır. `AZURE_OPENAI_API_KEY` ve `AZURE_OPENAI_ENDPOINT` gibi ortam değişkenlerini okumak için bunu kullanabilirsiniz.

## Metin oluşturma

Metin oluşturmanın yolu `responses.create` metodunu kullanmaktır. İşte bir örnek:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yukarıdaki kodda, bir yanıt nesnesi oluşturup kullanmak istediğimiz modeli ve istemi geçiyoruz. Sonra oluşturulan metni yazdırıyoruz.

## Alıştırma - bir tarif oluşturucu yapın

Şimdi bir senaryo oynadığımıza göre, gösterilen senaryoya uygun kod yazalım. Bunu yapmak için şu adımları izleyin:

1. Mevcut *app.py* dosyasını başlangıç noktası olarak kullanın
1. `prompt` değişkenini bulun ve kodunu aşağıdaki gibi değiştirin:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

To further improve it, we want to add the following:

- **Filter out ingredients**. We want to be able to filter out ingredients we don't like or are allergic to. To accomplish this change, we can edit our existing prompt and add a filter condition to the end of it like so:

    ```python

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```
