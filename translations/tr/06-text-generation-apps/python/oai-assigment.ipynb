## Metin oluşturma uygulaması nedir?

Normalde bir uygulama oluşturduğunuzda şu gibi bir arayüzü olur:

- Komut tabanlı. Konsol uygulamaları, bir komut yazdığınız ve görevi yerine getirdiği tipik uygulamalardır. Örneğin, `git` komut tabanlı bir uygulamadır.
- Kullanıcı arayüzü (UI). Bazı uygulamaların butonlara tıkladığınız, metin girdiğiniz, seçenekler seçtiğiniz grafiksel kullanıcı arayüzleri (GUI) vardır.

### Konsol ve UI uygulamaları sınırlıdır

Yazdığınız komutları alabilen komut tabanlı uygulamayla karşılaştırın:

- **Sınırlıdır**. Her komutu yazamazsınız, sadece uygulamanın desteklediği komutları yazabilirsiniz.
- **Dile özgüdür**. Bazı uygulamalar birçok dili destekler, ancak varsayılan olarak uygulama belirli bir dile göre oluşturulmuştur, dil desteği ekleyebilmeniz mümkün olsa bile.

### Metin oluşturma uygulamalarının faydaları

Peki, metin oluşturma uygulaması neden farklıdır?

Metin oluşturma uygulamasında daha fazla esnekliğe sahipsiniz, belirli bir komut seti veya belirli bir giriş dili ile sınırlı değilsiniz. Bunun yerine uygulama ile doğal dili kullanarak etkileşime girebilirsiniz. Başka bir fayda ise, zaten çok geniş bir metin koleksiyonu üzerinde eğitilmiş bir veri kaynağıyla etkileşimde olmanızdır; geleneksel bir uygulama ise yalnızca veritabanındaki verilerle sınırlı olabilir.

### Metin oluşturma uygulaması ile neler yapabilirim?

Birçok şey yapabilirsiniz. Örneğin:

- **Bir sohbet botu**. Şirketiniz ve ürünleriyle ilgili soruları yanıtlayan bir sohbet botu iyi bir eşleşme olabilir.
- **Yardımcı**. LLM'ler metin özetleme, metinden çıkarım yapma, özgeçmiş gibi metin üretme gibi görevlerde çok iyidir.
- **Kod asistanı**. Kullandığınız dil modeline bağlı olarak, kod yazmanıza yardımcı bir kod asistanı yapabilirsiniz. Örneğin, GitHub Copilot veya ChatGPT gibi ürünleri kullanarak kod yazmanıza destek alabilirsiniz.

## Metin oluşturun

Metin oluşturmanın yolu `responses.create` metodunu kullanmaktır. İşte bir örnek:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yukarıdaki kodda bir yanıt nesnesi oluşturuyoruz ve kullanmak istediğimiz modeli ve prompt'u veriyoruz. Sonra oluşturulan metni yazdırıyoruz.

### Sohbet yanıtları

Şimdiye kadar, tek bir prompt'tan nasıl metin oluşturacağımızı gördünüz. Yanıtlar API'si sohbet botları için de çok uygundur - konuşmayı oluşturmak için `input` olarak bir mesaj listesi sağlarsınız. İşte bunun bir örneği:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Bu işlevselliğin daha fazlasını gelecek bir bölümde göreceksiniz.

## Alıştırma - bir tarif oluşturucu yapın

Şimdi bir senaryo oynadığımıza göre, gösterilen senaryoya uygun kod yazalım. Bunu yapmak için aşağıdaki adımları izleyin:

1. Başlangıç noktası olarak mevcut *app.py* dosyasını kullanın
1. `prompt` değişkenini bulun ve kodunu aşağıdakilerle değiştirin:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. Kodu aşağıdaki şekilde değiştirelim:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

        ```python
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Yeni bir istek yapıyoruz, ayrıca ilk promptta istediğimiz token sayısını göz önünde bulundurarak bu sefer `max_output_tokens` 1200 olarak belirliyoruz.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Bu kodu çalıştırdığımızda aşağıdaki çıktıyı elde ediyoruz:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.
        
        -Apple and flour pancakes: 1 cup flour, 1/2 tsp baking powder, 1/2 tsp baking soda, 1/4 tsp salt, 1 tbsp sugar, 1 egg, 1 cup buttermilk or sour milk, 1/4 cup melted butter, 1 Granny Smith apple, peeled and grated
        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **Token uzunluğuyla ilgili bir not**. Metin üretmek için kaç token gerektiğini göz önünde bulundurmalıyız. Tokenler maliyetlidir, bu yüzden mümkün olduğunca token sayısını ekonomik kullanmak iyidir. Örneğin promptu daha az token kullanacak şekilde formüle edebilir miyiz?

   Kullanılan token sayısını değiştirmek için `max_output_tokens` parametresini kullanabilirsiniz. Örneğin 100 token kullanmak istiyorsanız şu şekilde yaparsınız:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Sıcaklıkla oynama**. Sıcaklıktan şimdiye kadar bahsetmedik ama bu programımızın performansı için önemli bir bağlamdır. Sıcaklık değeri ne kadar yüksek olursa çıktı o kadar rastgele olur. Tersi durumda, sıcaklık değeri düşükse çıktı daha öngörülebilir olur. Çıktınızda varyasyon isteyip istemediğinizi düşünün.

   Sıcaklığı değiştirmek için `temperature` parametresini kullanabilirsiniz. Örneğin 0.5 sıcaklık kullanmak isterseniz şöyle yaparsınız:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Not, 1.0'a ne kadar yakınsa çıktı o kadar çeşitli olur.
