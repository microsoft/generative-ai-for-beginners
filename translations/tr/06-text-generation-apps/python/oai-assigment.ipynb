## Metin oluşturma uygulaması nedir?

Normalde bir uygulama geliştirdiğinizde, aşağıdakilere benzer bir arayüzü olur:

- Komut tabanlı. Konsol uygulamaları, bir komut yazdığınız ve görev gerçekleştiren tipik uygulamalardır. Örneğin, `git` komut tabanlı bir uygulamadır.
- Kullanıcı arayüzü (UI). Bazı uygulamalar, düğmelere tıklamanız, metin girmeniz, seçenekler seçmeniz ve daha fazlasını yapabileceğiniz grafik kullanıcı arayüzlerine (GUI) sahiptir.

### Konsol ve UI uygulamaları sınırlıdır

Bir komut yazdığınız komut tabanlı bir uygulama ile karşılaştırın: 

- **Sınırlıdır**. Herhangi bir komut yazamazsınız, yalnızca uygulamanın desteklediği komutları yazabilirsiniz.
- **Dile özgüdür**. Bazı uygulamalar birçok dili destekler, ancak varsayılan olarak uygulama belirli bir dil için geliştirilmiştir, birkaç dil desteği ekleyebilmenize rağmen.

### Metin oluşturma uygulamalarının avantajları

Metin oluşturma uygulaması nasıl farklı?

Metin oluşturma uygulamasında daha fazla esnekliğe sahipsiniz, belirli komutlar veya özel bir giriş dili ile sınırlı değilsiniz. Bunun yerine uygulama ile etkileşimde bulunmak için doğal dil kullanabilirsiniz. Ayrıca, zaten geniş bir bilgi kümesi üzerinde eğitilmiş bir veri kaynağı ile etkileşimde bulunuyor olmanız bir avantajdır, geleneksel bir uygulama bir veritabanıyla sınırlı olabilir.

### Metin oluşturma uygulaması ile neler geliştirebilirim?

Birçok şey geliştirebilirsiniz. Örneğin:

- **Bir sohbet botu**. Şirketiniz ve ürünleri gibi konular hakkında soruları yanıtlayan bir sohbet botu iyi bir uyum olabilir.
- **Yardımcı**. Büyük dil modelleri (LLM) metin özetleme, metinden çıkarım yapma, özgeçmiş gibi metin üretme gibi konularda iyidir.
- **Kod asistanı**. Kullandığınız dil modeline bağlı olarak, size kod yazmada yardımcı olacak bir kod asistanı geliştirebilirsiniz. Örneğin, GitHub Copilot ve ChatGPT gibi ürünleri kod yazmanıza yardımcı olmak için kullanabilirsiniz.

> [!NOTE]

## Metin oluştur

Metin oluşturmanın yolu `responses.create` yöntemini kullanmaktır. İşte bir örnek:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Yukarıdaki kodda, bir yanıt nesnesi oluştururuz ve kullanmak istediğimiz modeli ile komut istemini iletiriz. Sonra oluşturulan metni yazdırırız.

### Sohbet yanıtları

Şimdiye kadar, tek bir komut isteminden nasıl metin oluşturabileceğimizi gördünüz. Responses API, sohbet botları için de uygundur - bir konuşmayı oluşturmak için `input` olarak bir mesajlar listesi sağlarsınız. İşte kullanıma bir örnek:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Bu işlevsellik hakkında daha fazlası gelecek bir bölümde.

## Alıştırma - bir tarif oluşturucu yap

Artık bir senaryoyu oynadığımıza göre, gösterilen senaryoya uygun kod yazalım. Bunu yapmak için şu adımları izleyin:

1. Mevcut *app.py* dosyasını başlangıç noktası olarak kullanın
1. `prompt` değişkenini bulun ve kodunu aşağıdaki gibi değiştirin:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Şu noktaları not edin:

- İlk istemin sonucunu yeni isteme ekleyerek yeni bir istem oluşturuyoruz:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

- Yeni bir istek yapıyoruz, fakat aynı zamanda ilk istemde istediğimiz token sayısını da dikkate alıyoruz, bu yüzden bu sefer `max_output_tokens` 1200 olarak belirleniyor.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

Bu kodu çalıştırdığınızda şu çıktıya ulaşıyoruz:

        ```output
        No of recipes (for example, 5: 2
        List of ingredients (for example, chicken, potatoes, and carrots: apple,flour
        Filter (for example, vegetarian, vegan, or gluten-free: sugar
        Recipes:
         or milk.

        -Apple fritters: 1-1/2 cups flour, 1 tsp baking powder, 1/4 tsp salt, 1/4 tsp baking soda, 1/4 tsp nutmeg, 1/4 tsp cinnamon, 1/4 tsp allspice, 1/4 cup sugar, 1/4 cup vegetable shortening, 1/4 cup milk, 1 egg, 2 cups shredded, peeled apples
        Shopping list:
         -Flour, baking powder, baking soda, salt, sugar, egg, buttermilk, butter, apple, nutmeg, cinnamon, allspice 
        ```
        
- **Token uzunluğu hakkında bir not**. Üretmek istediğimiz metin için kaç token gerektiğini düşünmeliyiz. Tokenler maliyetlidir, bu yüzden mümkün olduğunda token sayısını ekonomik kullanmalıyız. Örneğin, istemi daha az token kullanacak şekilde ifade edebilir miyiz?

   Kullanılan token sayısını değiştirmek için `max_output_tokens` parametresini kullanabilirsiniz. Örneğin, 100 token kullanmak istiyorsanız şu şekilde yaparsınız:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Sıcaklık ile deney yapmak**. Sıcaklık daha önce bahsetmediğimiz ama programımızın performansı için önemli bir bağlamdır. Sıcaklık değeri arttıkça çıktı daha rastgele olur. Tam tersine sıcaklık değeri düşükse çıktı daha tahmin edilebilir olur. Çıktınızda varyasyon isteyip istemediğinizi düşünün.

   Sıcaklığı değiştirmek için `temperature` parametresini kullanabilirsiniz. Örneğin, 0.5 sıcaklık kullanmak istiyorsanız şöyle yaparsınız:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Not, 1.0'a ne kadar yaklaşırsanız çıktı o kadar çeşitli olur.
