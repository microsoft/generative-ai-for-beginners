## Generiranje besedila

Za generiranje besedila uporabimo metodo `responses.create`. Tukaj je primer:

```python

```

V zgornji kodi ustvarimo odgovor in posredujemo model, ki ga želimo uporabiti, ter poziv. Nato izpišemo generirano besedilo.

### Odgovori iz klepeta

Do zdaj ste videli, kako lahko generiramo besedilo iz enega poziva. API Responses je prav tako primeren za chatbote – podate seznam sporočil kot `input` za gradnjo pogovora. Tukaj je primer uporabe:

```python

  api_key = os.environ['OPENAI_API_KEY']
  )

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)

```

## Naloga - izdelava generatorja receptov

Zdaj, ko smo odigrali scenarij, napišimo kodo, ki ustreza prikazanemu scenariju. Za to sledite naslednjim korakom:

1. Uporabite obstoječo datoteko *app.py* kot izhodišče
1. Poiščite spremenljivko `prompt` in spremenite njeno kodo na naslednjo:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Ustvarjanje nakupovalnega seznama**. Želimo ustvariti nakupovalni seznam, ob upoštevanju, kaj že imamo doma.

    Za to funkcionalnost lahko poskusimo rešiti vse v enem pozivu ali pa jo razdelimo na dva poziva. Poskusimo slednji pristop. Tukaj predlagamo dodati dodaten poziv, vendar da to deluje, moramo rezultat prvega poziva dodati kot kontekst drugemu pozivu.

    Poiščite del kode, ki izpisuje rezultat prvega poziva, in dodajte spodnjo kodo:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # natisni odgovor
    print("Shopping list:")
    print(response.output_text)
    ```

    Upoštevajte naslednje:

    - Konstruiramo nov poziv tako, da dodamo rezultat prvega poziva v novi poziv:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Naredimo nov zahtevek, vendar upoštevamo število tokenov, za katere smo zaprosili v prvem pozivu, tako da tokrat rečemo, da je `max_output_tokens` 1200.

