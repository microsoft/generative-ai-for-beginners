## Ano ang app para sa pagbuo ng teksto?

Karaniwan kapag gumagawa ka ng app, may uri itong interface tulad ng mga sumusunod:

- Batay sa command. Ang mga console apps ay mga tipikal na app kung saan nagta-type ka ng command at isinasagawa nito ang isang gawain. Halimbawa, ang `git` ay isang command-based app.
- User interface (UI). Ang ilang app ay may graphical user interfaces (GUI) kung saan nagki-click ka ng mga button, naglalagay ng teksto, pumipili ng mga opsyon at iba pa.

### Limitado ang Console at UI apps

Ikompara ito sa isang command-based app kung saan nagta-type ka ng command:

- **Limitado ito**. Hindi ka basta-basta makakapag-type ng anumang command, tanging ang mga sinusuportahan ng app lang.
- **Tiyak sa wika**. Ang ilang app ay sumusuporta sa maraming wika, pero sa default ay ginawa ang app para sa isang partikular na wika, kahit na maaari kang magdagdag ng suporta para sa ibang mga wika.

### Mga Benepisyo ng mga app para sa pagbuo ng teksto

Ano naman ang pagkakaiba ng app para sa pagbuo ng teksto?

Sa isang app para sa pagbuo ng teksto, mas malawak ang iyong kakayahan, hindi ka limitado sa isang set ng mga command o isang tiyak na wikang input. Sa halip, maaari kang gumamit ng natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa pinanggagalingan ng datos na na-train sa malawak na korpus ng impormasyon, samantalang ang tradisyonal na app ay maaaring limitado lamang sa kung ano ang nasa isang database.

### Ano ang pwede kong gawin gamit ang app para sa pagbuo ng teksto?

Maraming bagay ang maaari mong gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nito ay maaaring maging angkop.
- **Taga-tulong**. Ang mga LLM ay mahusay sa mga bagay tulad ng pagsasummarize ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng teksto tulad ng mga resume at iba pa.
- **Katulong sa pag-cocode**. Depende sa language model na gagamitin mo, maaari kang gumawa ng katulong sa pag-cocode na tumutulong sa pagsusulat ng code. Halimbawa, maaari kang gumamit ng produktong tulad ng GitHub Copilot pati na rin ang ChatGPT para makatulong sa pagsusulat ng code.

> [!NOTE]

## Gumawa ng teksto

Ang paraan para gumawa ng teksto ay gamitin ang `responses.create` na pamamaraan. Narito ang halimbawa:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Sa kodigo sa itaas, gumagawa tayo ng response na bagay at pinapasa ang modelo na gusto nating gamitin at ang prompt. Saka natin piniprint ang nabuo na teksto.

### Mga sagot sa chat

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```
