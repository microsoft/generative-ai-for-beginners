## Ano ang text generation app?

Karaniwan kapag gumagawa ka ng app, ito ay may isang uri ng interface tulad ng sumusunod:

- Batay sa command. Ang mga console apps ay karaniwang apps kung saan nagta-type ka ng command at ito ay nagsasagawa ng isang gawain. Halimbawa, ang `git` ay isang command-based app.
- User interface (UI). Ang ilang apps ay may graphical user interfaces (GUIs) kung saan nagki-click ka ng mga button, nag-iinput ng teksto, pumipili ng mga opsyon at marami pa.

### Limitado ang console at UI apps

Ihambing ito sa isang command-based app kung saan nagta-type ka ng command:

- **Limitado ito**. Hindi ka pwedeng mag-type ng anumang command, kundi ang mga sinusuportahan ng app lang.
- **Nakatutok sa partikular na wika**. Ang ilang apps ay sumusuporta sa maraming wika, ngunit default, ang app ay ginawa para sa isang partikular na wika, kahit na pwede kang magdagdag ng suporta para sa iba pang wika.

### Mga benepisyo ng text generation apps

Paano naman naiiba ang text generation app?

Sa isang text generation app, mas may kalayaan ka, hindi ka limitado sa hanay ng mga command o partikular na input na wika. Sa halip, maaari kang gumamit ng natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang data source na na-train sa napakalawak na koleksyon ng impormasyon, samantalang ang tradisyunal na app ay maaaring limitado lang sa database na iyon.

### Ano ang pwedeng gawin gamit ang text generation app?

Maraming bagay ang pwedeng gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng kumpanya mo at mga produkto nito ay maaaring maging isang magandang halimbawa.
- **Taga-tulong**. Magaling ang LLMs sa mga bagay tulad ng pagbubuod ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng mga teksto tulad ng resume at marami pa.
- **Katulong sa code**. Depende sa ginagamit na language model, maaari kang gumawa ng code assistant na tumutulong sa pagsusulat ng code. Halimbawa, maaari mong gamitin ang produktong GitHub Copilot pati na rin ang ChatGPT para tulungan kang sumulat ng code.

### I-setup ang configuration ng Azure

Kung gumagamit ka ng Azure OpenAI, ganito ang pag-set up ng configuration. Ang Responses API ay pinagsisilbihan mula sa Azure OpenAI **v1 endpoint**, kaya itinuturo natin ang `OpenAI` client sa `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Sa itaas, itinatakda natin ang mga sumusunod:

- `api_key`, ito ang iyong API key na makikita sa Azure Portal.
- `base_url`, ito ang iyong Azure OpenAI endpoint na may `/openai/v1/` na idinagdag. Makikita mo ang endpoint sa Azure Portal malapit sa iyong API key. Ang paggamit ng v1 endpoint ay nangangahulugan na hindi mo na kailangang magpasa ng `api_version`.
- `deployment`, ito ang pangalan ng model deployment na ginawa mo sa Foundry portal.

> [!NOTE]
> Ang `os.environ` ay isang mapping na nagbabasa ng environment variables. Magagamit mo ito para basahin ang mga environment variable tulad ng `AZURE_OPENAI_API_KEY` at `AZURE_OPENAI_ENDPOINT`.

## Mag-generate ng teksto

Ang paraan para mag-generate ng teksto ay gamit ang `responses.create` method. Narito ang isang halimbawa:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Sa code sa itaas, gumawa tayo ng response object at ipinasa ang model na gagamitin at ang prompt. Pagkatapos, ipinrint natin ang generated na teksto.

### Mga tugon sa chat

Ang Responses API ay angkop para sa parehong single-turn text generation at multi-turn chatbots - nagbibigay ka lang ng mas maraming mensahe sa `input` list upang bumuo ng pag-uusap:

```python
client = OpenAI(
```

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```
