# Gumawa ng mga app para sa pagbuo ng teksto

Nakita mo na sa bahaging ito ng kurikulum na may mga pangunahing konsepto tulad ng mga prompt at pati na rin ang isang buong disiplina na tinatawag na "prompt engineering". Maraming tool na maaari mong makausap tulad ng ChatGPT, Office 365, Microsoft Power Platform at iba pa, na sumusuporta sa'yo sa paggamit ng mga prompt para makamit ang isang bagay.

Para makadagdag ng ganitong karanasan sa isang app, kailangan mong maunawaan ang mga konsepto tulad ng mga prompt, mga kompletong sagot (completions) at pumili ng library na gagamitin. Iyan mismo ang matututunan mo sa kabanatang ito.

## Panimula

Sa kabanatang ito, iyong:

- Matutunan ang tungkol sa openai library at ang mga pangunahing konsepto nito.
- Gumawa ng app para sa pagbuo ng teksto gamit ang openai.
- Maunawaan kung paano gamitin ang mga konsepto tulad ng prompt, temperatura, at mga tokens para gumawa ng app sa pagbuo ng teksto.

## Mga layunin sa pag-aaral

Sa pagtatapos ng araling ito, magagawa mong:

- Ipaliwanag kung ano ang isang app para sa pagbuo ng teksto.
- Gumawa ng app para sa pagbuo ng teksto gamit ang openai.
- I-configure ang iyong app upang gumamit ng mas marami o mas kakaunting mga tokens at baguhin din ang temperatura, para sa iba't ibang resulta.

## Ano ang app para sa pagbuo ng teksto?

Karaniwang kapag gumagawa ka ng app, ito ay may uri ng interface tulad ng sumusunod:

- Batay sa command. Ang mga console app ay mga karaniwang app na nagta-type ka ng command at ito ay nagsasagawa ng gawain. Halimbawa, ang `git` ay isang command-based app.
- Interface ng gumagamit (UI). Ang ilang mga app ay may graphical user interfaces (GUI) kung saan nagki-click ka ng mga button, naglalagay ng teksto, pumipili ng opsyon, atbp.

### Limitado ang mga Console at UI app

Ihambing ito sa isang command-based na app kung saan nagta-type ka ng command:

- **Ito ay limitado**. Hindi ka basta-basta makakapag-type ng anumang command, kundi yung mga sinusuportahan ng app.
- **Nakatukoy sa wika**. Ang ilang app ay sumusuporta sa maraming wika, pero sa simula ay ginawa ang app para sa isang partikular na wika, kahit na maaari kang magdagdag ng suporta sa iba pang wika.

### Mga benepisyo ng mga app para sa pagbuo ng teksto

Paano naman naiiba ang app para sa pagbuo ng teksto?

Sa isang app para sa pagbuo ng teksto, mayroon kang higit na kalayaan, hindi ka limitado sa isang hanay ng mga command o isang partikular na input na wika. Sa halip, maaari mong gamitin ang natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang pinagmumulan ng datos na tinuruan gamit ang napakalawak na hanay ng impormasyon, samantalang ang isang tradisyunal na app ay maaaring limitado lamang sa laman ng database.

### Ano ang maaari kong gawin gamit ang isang app sa pagbuo ng teksto?

Maraming bagay ang maaari mong gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nito, ay maaaring magandang halimbawa.
- **Tumutulong**. Mahusay ang LLM sa mga bagay tulad ng pagsasabi ng buod ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng teksto tulad ng mga resume at iba pa.
- **Asistenteng pang-kodigo**. Depende sa language model na ginagamit mo, maaari kang gumawa ng code assistant na tumutulong sa pagsulat ng kodigo. Halimbawa, maaari mong gamitin ang mga produkto tulad ng GitHub Copilot pati na rin ang ChatGPT para tulungan kang magsulat ng code.

## Ehersisyo - gumawa ng recipe generator

Ngayon na naglaro tayo ng isang senaryo, sumulat tayo ng code upang tumugma sa ipinakitang senaryo. Upang gawin ito, sundin ang mga hakbang na ito:

1. Gamitin ang umiiral na file bilang panimulang punto
1. Gumawa ng `prompt` variable at baguhin ang sample code tulad ng nasa ibaba:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Maganda, tingnan natin kung paano natin mapapabuti ang mga bagay. Upang mapabuti ang mga bagay, nais nating matiyak na ang code ay flexible, kaya ang mga sangkap at bilang ng mga resipe ay maaaring mapabuti at mabago.

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

### Pagbutihin sa pamamagitan ng pagdaragdag ng filter at listahan ng pamimili

Ngayon ay mayroon na tayong gumaganang app na kayang gumawa ng mga recipe at ito ay flexible dahil umaasa ito sa mga input mula sa user, kapwa sa bilang ng mga recipe pati na rin ang mga sangkap na ginamit.

Para lalo pa itong pagbutihin, nais nating idagdag ang mga sumusunod:

- **I-filter ang mga sangkap**. Nais nating magkaroon ng kakayahan na i-filter ang mga sangkap na hindi namin gusto o may allergy kami. Para makamit ang pagbabagong ito, pwede nating i-edit ang ating kasalukuyang prompt at magdagdag ng filter condition sa dulo nito tulad nito:

    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")
