## Ano ang text generation app?

Karaniwan kapag gumagawa ka ng app, may interface ito tulad ng mga sumusunod:

- Command-based. Ang mga console app ay karaniwang app kung saan nagta-type ka ng command at gagawin nito ang isang gawain. Halimbawa, ang `git` ay isang command-based app.
- User interface (UI). May mga app na may graphical user interfaces (GUIs) kung saan nagki-click ka ng mga button, naglalagay ng text, pumipili ng options at iba pa.

### Limitado ang Console at UI apps

Ihambing ito sa command-based app kung saan nagta-type ka ng command:

- **Limitado ito**. Hindi ka basta-basta makakapag-type ng kahit anong command, kundi yung mga sinusuportahan lang ng app.
- **Partikular sa wika**. May mga app na sumusuporta sa maraming wika, pero kadalasan, ang app ay ginawa para sa isang partikular na wika, kahit puwede kang magdagdag ng suporta sa iba pang wika.

### Mga benepisyo ng text generation apps

Ano ang pinagkaiba ng text generation app?

Sa text generation app, mas malaya ka, hindi ka limitado sa set ng commands o isang partikular na input na wika. Sa halip, puwede kang gumamit ng natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang data source na na-train sa napakaraming impormasyon, samantalang ang tradisyonal na app ay maaaring limitado lang sa laman ng database.

### Ano ang puwede kong gawin gamit ang text generation app?

Maraming bagay ang puwede mong gawin. Halimbawa:

- **Chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nito ay magandang halimbawa.
- **Helper**. Magaling ang LLMs sa mga bagay tulad ng pag-summarize ng text, pagkuha ng insights mula sa text, paggawa ng mga text tulad ng resume at iba pa.
- **Code assistant**. Depende sa language model na gagamitin mo, puwede kang gumawa ng code assistant na tutulong sa iyo magsulat ng code. Halimbawa, puwede mong gamitin ang produkto tulad ng GitHub Copilot pati na rin ang ChatGPT para tumulong sa pagsusulat ng code.

## Paano ako makakapagsimula?

Kailangan mong maghanap ng paraan para makipag-integrate sa isang LLM na karaniwang may dalawang paraan:

- Gumamit ng API. Dito, gumagawa ka ng web requests gamit ang iyong prompt at makakakuha ka ng generated na text pabalik.
- Gumamit ng library. Ang mga library ay tumutulong para gawing mas madali ang paggamit ng API calls.

## Ehersisyo - gumawa ng recipe generator

Ngayon na naipakita na natin ang isang senaryo, magsulat naman tayo ng code na tumutugma sa ipinakitang senaryo. Sundin ang mga hakbang na ito:

1. Gamitin ang kasalukuyang file bilang panimulang punto
1. Gumawa ng variable na `prompt` at baguhin ang sample code gaya ng nasa ibaba:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Ayos, tingnan natin kung paano pa natin mapapaganda ang mga bagay. Para mapabuti pa ito, gusto nating siguraduhin na flexible ang code, para madali nating mapalitan o madagdagan ang mga sangkap at bilang ng mga recipe.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

    Hanapin ang bahagi ng code na nagpi-print ng resulta mula sa unang prompt at idagdag ang sumusunod na code sa ibaba:

    old_prompt_result = response.choices[0].message.content
