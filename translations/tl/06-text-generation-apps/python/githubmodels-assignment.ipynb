# Bumuo ng mga app para sa pagbuo ng teksto

Nakita mo na sa kurikulum na ito na may mga pangunahing konsepto tulad ng mga prompt at pati na rin ang isang buong disiplina na tinatawag na "prompt engineering". Maraming mga tool na maaari mong makausap tulad ng ChatGPT, Office 365, Microsoft Power Platform at marami pa, na sumusuporta sa iyo sa paggamit ng mga prompt upang makamit ang isang bagay.

Para maidagdag mo ang ganitong karanasan sa isang app, kailangan mong maunawaan ang mga konsepto tulad ng mga prompt, completions at pumili ng isang library na gagamitin. Iyan mismo ang matututuhan mo sa kabanatang ito.

## Mga layunin sa pagkatuto

Sa katapusan ng araling ito, magagawa mong:

- Ipaliwanag kung ano ang isang text generation app.
- Gumawa ng isang text generation app gamit ang openai.
- I-configure ang iyong app gamit ang mas marami o mas kaunting tokens at baguhin ang temperature, para sa iba’t ibang resulta.

## Ano ang text generation app?

Karaniwan kapag bumubuo ka ng app ito ay may uri ng interface tulad ng sumusunod:

- Batay sa command. Ang mga console app ay mga tipikal na app kung saan nagta-type ka ng command at isinasagawa nito ang isang gawain. Halimbawa, ang `git` ay isang command-based app.
- User interface (UI). Ang ilang mga app ay may graphical user interface (GUIs) kung saan nagki-click ka ng mga button, nag-iinput ng teksto, pumipili ng mga opsyon at iba pa.

### May hangganan ang Console at UI apps

Ihambing ito sa isang command-based app kung saan nagta-type ka ng command:

- **May limitasyon ito**. Hindi ka maaaring mag-type ng kahit anong command, tanging ang mga sinusuportahan lamang ng app.
- **Tiyak sa wika**. Ang ilang app ay sumusuporta ng maraming wika, ngunit sa default ang app ay ginawa para sa isang partikular na wika, kahit na maaari kang magdagdag ng suporta para sa iba pang mga wika.

### Mga benepisyo ng text generation apps

Paano nga ba naiiba ang text generation app?

Sa isang text generation app, may mas malaking kalayaan ka, hindi ka limitado sa set ng mga command o mga espesipikong wika para sa input. Sa halip, maaari mong gamitin ang natural na wika para makipag-interact sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang data source na tinuruan gamit ang napakalawak na koleksyon ng impormasyon, samantalang ang tradisyunal na app ay limitado sa laman ng database.

### Ano ang maaari kong buuin gamit ang text generation app?

Maraming bagay ang maaari mong gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nito ay maaaring maging angkop.
- **Tumutulong**. Mahusay ang mga LLM sa mga bagay tulad ng pagsasummarize ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng mga dokumento tulad ng resume at iba pa.
- **Katulong sa code**. Depende sa language model na gagamitin mo, maaari kang gumawa ng katulong sa code na tutulong sa pagsusulat ng code. Halimbawa, maaari mong gamitin ang produkto tulad ng GitHub Copilot pati na rin ang ChatGPT upang makatulong sa pagsusulat ng code.

## Ehersisyo - gumawa ng isang generator ng recipe

Ngayon na naipakita na natin ang isang senaryo, sumulat tayo ng code na tumutugma sa naipakitang senaryo. Upang gawin ito, sundin ang mga hakbang na ito:

1. Gamitin ang umiiral na file bilang panimulang punto
1. Gumawa ng variable na `prompt` at baguhin ang sample code tulad ng nasa ibaba:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Magaling, tingnan natin kung paano natin mapapabuti ang mga bagay. Upang mapabuti ang mga bagay, nais nating siguraduhin na ang code ay flexible, upang ang mga sangkap at bilang ng mga recipe ay maaaring mapabuti at mabago.


1. Baguhin natin ang code sa sumusunod na paraan:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

### Pahusayin sa pamamagitan ng pagdagdag ng filter at shopping list
