## Ano ang text generation app?

Karaniwan kapag gumagawa ka ng app ito ay may isang uri ng interface tulad ng sumusunod:

- Batay sa command. Karaniwang ang mga console apps ay mga uri ng app kung saan nagta-type ka ng command at ito ay nagsasagawa ng isang gawain. Halimbawa, `git` ay isang command-based app.
- User interface (UI). May ilang app na may graphical user interfaces (GUIs) kung saan magki-click ka ng mga button, maglalagay ng teksto, pipili ng mga opsyon at iba pa.

### Limitado ang Console at UI apps

Ihambing ito sa command-based app kung saan nagta-type ka ng command:

- **Limitado ito**. Hindi ka basta-basta makakagawa ng anumang command, kundi lamang ang mga sinusuportahan ng app.
- **Espesipiko sa wika**. May ilang app na sumusuporta sa maraming wika, ngunit sa default ang app ay ginawa para sa isang partikular na wika, kahit na maaari kang magdagdag pa ng suporta sa wika.

### Mga benepisyo ng text generation apps

Paano ba naiiba ang text generation app?

Sa text generation app, mas malaki ang flexibility, hindi ka limitado sa hanay ng mga command o partikular na wika ng input. Sa halip, maaari kang gumamit ng natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang data source na na-train sa napakalawak na koleksyon ng impormasyon, samantalang ang tradisyunal na app ay maaaring limitado sa nilalaman ng isang database.

### Ano ang maaari kong gawin gamit ang text generation app?

Maraming bagay ang maaari mong gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nitong maaaring mahusay na akma.
- **Tulong**. Mahusay ang LLMs sa mga bagay tulad ng pagbubuod ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng teksto tulad ng mga resume at iba pa.
- **Katulong sa code**. Depende sa language model na gagamitin mo, maaari kang gumawa ng code assistant na tumutulong sa pagsulat ng code. Halimbawa, maaari mong gamitin ang mga produkto tulad ng GitHub Copilot pati na rin ang ChatGPT upang makatulong sa pagsulat ng code.

### Mga chat responses

Sa ngayon, nakita mo kung paano tayo makakagawa ng teksto mula sa isang prompt lang. Ang Responses API ay angkop din para sa mga chatbot - magbibigay ka ng listahan ng mga mensahe bilang `input` upang bumuo ng usapan. Heto ang halimbawa ng paggamit nito:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)

```

## Ehersisyo - gumawa ng recipe generator

Ngayon na naipakita na natin ang isang senaryo, sumulat tayo ng code na tutugma sa ipinakitang senaryo. Gawin ito sa pamamagitan ng pagsunod sa mga hakbang:

1. Gamitin ang umiiral na *app.py* file bilang panimulang punto
1. Hanapin ang variable na `prompt` at baguhin ang code nito sa sumusunod:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

May bayad ang mga token, kaya kung maaari, dapat tayong maging matipid sa dami ng token na ginagamit. Halimbawa, maaari bang gawing mas maikli ang prompt upang gumamit ng mas kaunting token?

   Upang baguhin ang dami ng token na ginagamit, maaari mong gamitin ang parameter na `max_output_tokens`. Halimbawa, kung gusto mong gamitin ang 100 token, gagawin mo ito:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Pagsubok sa temperatura**. Ang temperatura ay isang bagay na hindi pa natin nabanggit ngunit mahalaga sa konteksto kung paano gumagana ang ating programa. Kapag mas mataas ang halaga ng temperatura, mas random ang magiging output. Sa kabilang banda, kapag mas mababa ang temperatura, mas prediktable ang output. Isipin kung gusto mo ba ng variation sa iyong output o hindi.

   Upang baguhin ang temperatura, gamitin mo ang parameter na `temperature`. Halimbawa, kung gusto mong gamitin ang temperatura na 0.5, gagawin mo ito:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Tandaan, kapag mas malapit sa 1.0, mas maraming variation ang output.
