## Ano ang text generation app?

Karaniwan kapag gumagawa ka ng app, may interface ito tulad ng mga sumusunod:

- Command-based. Ang mga console app ay karaniwang app kung saan nagta-type ka ng command at may ginagawa itong gawain. Halimbawa, ang `git` ay isang command-based app.
- User interface (UI). May mga app na may graphical user interfaces (GUIs) kung saan nagki-click ka ng mga button, naglalagay ng text, pumipili ng mga opsyon at iba pa.

### Limitado ang Console at UI apps

Ihambing ito sa command-based app kung saan nagta-type ka ng command:

- **Limitado ito**. Hindi ka basta-basta makakapag-type ng kahit anong command, kundi yung mga sinusuportahan lang ng app.
- **Partikular sa wika**. May mga app na sumusuporta sa maraming wika, pero kadalasan, ang app ay ginawa para sa isang partikular na wika, kahit pwede kang magdagdag ng suporta sa iba pang wika.

### Mga benepisyo ng text generation apps

Ano ang pinagkaiba ng text generation app?

Sa text generation app, mas malaya ka, hindi ka limitado sa set ng mga command o isang partikular na input na wika. Sa halip, pwede kang gumamit ng natural na wika para makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang data source na na-train sa napakaraming impormasyon, samantalang ang tradisyonal na app ay limitado lang sa kung anong laman ng database.

### Ano ang pwede kong gawin gamit ang text generation app?

Maraming bagay ang pwede mong gawin. Halimbawa:

- **Chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksa, tulad ng iyong kumpanya at mga produkto nito ay magandang halimbawa.
- **Helper**. Magaling ang LLMs sa mga bagay tulad ng pag-summarize ng teksto, pagkuha ng insights mula sa teksto, paggawa ng mga teksto tulad ng resume at iba pa.
- **Code assistant**. Depende sa language model na gagamitin mo, pwede kang gumawa ng code assistant na tutulong sa iyo magsulat ng code. Halimbawa, pwede mong gamitin ang produkto tulad ng GitHub Copilot pati na rin ang ChatGPT para tumulong sa pagsusulat ng code.

## Bumuo ng teksto

Ang paraan para bumuo ng teksto ay gamit ang `chat.completion` class. Heto ang halimbawa:

```python
prompt = "Complete the following: Once upon a time there was a"
```

Sa code sa itaas, gumagawa tayo ng completion object at pinapasa ang model na gusto nating gamitin at ang prompt. Pagkatapos, piniprint natin ang nabuo na teksto.

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=100)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, temperature=0.5)
    ```

## Pagsusulit sa Kaalaman

Ano ang ginagawa ng konseptong temperature?

1. Kinokontrol nito kung gaano ka-random ang output.
1. Kinokontrol nito kung gaano kalaki ang response.
1. Kinokontrol nito kung ilang tokens ang magagamit.

A: 1

Ano ang magandang paraan para mag-imbak ng mga sikreto tulad ng API keys?

1. Sa code.
1. Sa isang file.
1. Sa environment variables.

A: 3, dahil ang environment variables ay hindi naka-imbak sa code at pwedeng i-load mula sa code.
