# Gumawa ng mga app para sa pagbuo ng teksto

Para makapagdagdag ka ng ganitong karanasan sa isang app, kailangan mong maunawaan ang mga konsepto tulad ng mga prompt, mga compleksyon at pumili ng isang library na gagamitin. Iyan mismo ang iyong matututuhan sa kabanatang ito.

## Mga layunin sa pagkatuto

## Ano ang isang app para sa pagbuo ng teksto?

Karaniwan kapag gumawa ka ng app, mayroong ilang uri ng interface tulad ng sumusunod:

- Command-based. Ang mga console app ay mga tipikal na app kung saan nagtatype ka ng isang command at isinasagawa nito ang isang gawain. Halimbawa, ang `git` ay isang command-based na app.
- User interface (UI). Ang ilang mga app ay may graphical user interfaces (GUIs) kung saan pinipindot mo ang mga button, naglalagay ng teksto, pumipili ng mga opsyon at iba pa.

### Limitado ang mga Console at UI apps

Ihambing ito sa isang command-based na app kung saan nagtatype ka ng command:

- **Limitado ito**. Hindi ka basta-basta makakapag-type ng anumang command, tanging ang mga sinusuportahan ng app lang.
- **Naka-sentro sa wika**. Ang ilang mga app ay sumusuporta sa maraming mga wika, pero sa default, ang app ay ginawa para sa isang partikular na wika, kahit na maaari kang magdagdag ng suporta para sa iba pang mga wika.

### Mga benepisyo ng mga app para sa pagbuo ng teksto

Paano naman naiiba ang isang app para sa pagbuo ng teksto?

Sa isang app para sa pagbuo ng teksto, may mas maraming kalayaan ka, hindi ka limitado sa mga set ng mga command o isang tiyak na wika ng input. Sa halip, maaari mong gamitin ang natural na wika upang makipag-ugnayan sa app. Isa pang benepisyo ay dahil nakikipag-ugnayan ka na sa isang pinanggagalingan ng datos na na-train sa isang malawak na dami ng impormasyon, samantalang ang tradisyunal na app ay maaaring limitado lang sa kung ano ang nasa database.

### Ano ang maaari kong gawin gamit ang isang app para sa pagbuo ng teksto?

Maraming mga bagay ang maaari mong gawin. Halimbawa:

- **Isang chatbot**. Isang chatbot na sumasagot sa mga tanong tungkol sa mga paksang tulad ng iyong kumpanya at mga produkto nito ay maaaring maging magandang tugma.
- **Katulong**. Mahusay ang LLM sa mga bagay tulad ng pagsasaayos ng teksto, pagkuha ng mga insight mula sa teksto, paggawa ng teksto tulad ng mga resume at iba pa.
- **Katulong sa code**. Depende sa modelong pangwika na gagamitin mo, maaari kang gumawa ng katulong sa pag-cocode na tutulong sa iyo na magsulat ng code. Halimbawa, maaari mong gamitin ang mga produkto tulad ng GitHub Copilot pati na rin ang ChatGPT upang tulungan kang magsulat ng code.

## Gumawa ng teksto

Ang paraan para gumawa ng teksto ay gamitin ang `responses.create` method. Narito ang halimbawa:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Sa code sa itaas, gumawa tayo ng response object at ipinasa ang modelong gusto nating gamitin at ang prompt. Pagkatapos ay prinint natin ang na-generate na teksto.

### Mga tugon sa chat

Sa ngayon, nakita mo na kung paano tayo makakagawa ng teksto mula sa isang prompt lang. Ang Responses API ay angkop din para sa mga chatbot - nagbibigay ka ng listahan ng mga mensahe bilang `input` upang bumuo ng isang pag-uusap. Narito ang halimbawa ng paggamit nito:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Magbabahagi pa kami ng karagdagang impormasyon tungkol sa functionality na ito sa mga susunod na kabanata.

## Ehersisyo - gumawa ng isang recipe generator

Ngayon na nailarawan na natin ang isang senaryo, sumulat tayo ng code ayon sa ipinakitang senaryo. Upang gawin ito, sundin ang mga hakbang na ito:

1. Gamitin ang umiiral na *app.py* na file bilang panimulang punto
1. Hanapin ang variable na `prompt` at palitan ang code nito ng mga sumusunod:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. Baguhin natin ang code sa sumusunod na paraan:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Dito, nagmumungkahi tayong magdagdag ng karagdagang prompt, ngunit para magtrabaho ito, kailangan nating idagdag ang resulta ng naunang prompt bilang konteksto sa pangalawang prompt.

    Hanapin ang bahagi sa code na nagpi-print ng resulta mula sa unang prompt at idagdag ang sumusunod na code sa ibaba:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # iprint ang tugon
    print("Shopping list:")
    print(response.output_text)
    ```

    Pansinin ang mga sumusunod:

    - Gumagawa tayo ng bagong prompt sa pamamagitan ng pagdagdag ng resulta mula sa unang prompt sa bagong prompt:
    
        ```python

        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Gumagawa tayo ng bagong request, ngunit isinasaalang-alang din ang bilang ng mga token na hiniling natin sa unang prompt, kaya sa pagkakataong ito sinasabi nating `max_output_tokens` ay 1200.

- **Isang salita tungkol sa haba ng token**. Dapat nating isaalang-alang kung ilang token ang kailangan natin para mabuo ang tekstong nais natin. Ang mga token ay may bayad, kaya kapag maaari, dapat tayong magtipid sa paggamit ng mga token. Halimbawa, maaari bang baguhin ang prompt para gumamit ng mas kaunting token?

   Para baguhin ang bilang ng token na ginagamit, maaari mong gamitin ang parameter na `max_output_tokens`. Halimbawa, kung gusto mong gumamit ng 100 token, gagawin mo ito:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Pagsubok sa temperature**. Ang temperature ay isang bagay na hindi pa natin nababanggit ngunit mahalaga sa kung paano gumaganap ang ating programa. Kapag mas mataas ang halaga ng temperature, mas random ang output. Sa kabilang banda, kapag mas mababa ang temperature, mas predictable ang output. Isaalang-alang kung nais mong magkaroon ng pagkakaiba-iba sa iyong output o hindi.

   Para baguhin ang temperature, maaari mong gamitin ang parameter na `temperature`. Halimbawa, kung gusto mong gumamit ng temperatura na 0.5, gagawin mo ito:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Tandaan, kapag mas malapit sa 1.0, mas nagkakaroon ng pagkakaiba-iba ang output.
