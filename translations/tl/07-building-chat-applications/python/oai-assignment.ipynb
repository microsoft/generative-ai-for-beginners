### 2. I-import ang mga helper na library at i-instantiate ang mga kredensyal


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Paghahanap ng tamang modelo  
Ang mga modelo tulad ng GPT-4o at GPT-4o mini ay maaaring makaunawa at makabuo ng natural na wika, at available sa OpenAI platform na may iba't ibang antas ng kapangyarihan at bilis na angkop para sa iba't ibang mga gawain.


```python
# Select a general purpose chat model
model = "gpt-5-mini"
```

#### Paano kontrolin ang isang malaking language model  
"Sa lahat ng input sa isang malaking language model, ang pinaka-maimpluwensiya ay ang text prompt(1).

Maaaring bigyan ng prompt ang malalaking language model upang makabuo ng output sa ilang paraan:

Instruksyon: Sabihin sa modelo kung ano ang gusto mo
Pagkumpleto: Hikayatin ang modelo na kumpletuhin ang simula ng gusto mo
Demonstrasyon: Ipakita sa modelo kung ano ang gusto mo, gamit ang:
Ilang halimbawa sa prompt
Maraming daan o libo-libong mga halimbawa sa isang fine-tuning training dataset"



#### May tatlong pangunahing alituntunin sa paggawa ng prompt:

**Ipakita at sabihin**. Linawin kung ano ang gusto mo sa pamamagitan ng mga instruksyon, halimbawa, o kumbinasyon ng dalawa. Kung gusto mong ipang-uri sa modelo ang isang listahan ng mga item sa alpabetikong ayos o i-classify ang isang talata ayon sa sentimyento, ipakita sa kanya na iyon ang gusto mo.

**Magbigay ng de-kalidad na datos**. Kung sinusubukan mong bumuo ng isang classifier o gusto mong sundan ng modelo ang isang pattern, siguraduhing sapat ang mga halimbawa. Siguraduhing na-proofread ang iyong mga halimbawa — karaniwang matalino ang modelo upang makita ang mga karaniwang mali sa pagbantas at bibigyan ka ng tugon, ngunit maaari rin nitong isipin na ito ay sinasadya at makakaapekto sa tugon.

**Suriin ang iyong mga setting.** Kinokontrol ng mga setting na temperature at top_p kung gaano ka-determinado ang modelo sa pagbibigay ng tugon. Kung humihingi ka ng tugon kung saan iisa lamang ang tamang sagot, gusto mong pababain ang mga ito. Kung gusto mo ng mas sari-saring mga tugon, gusto mong itaas ang mga ito. Ang pinaka-karaniwang pagkakamali ng mga tao sa mga setting na ito ay ang pag-aakalang ito ay mga kontrol ng "talino" o "kreatibidad."


Source: https://learn.microsoft.com/azure/ai-foundry/openai/overview


### 5. Isumite!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Buodin ang Teksto  
#### Hamon  
Buodin ang teksto sa pamamagitan ng pagdagdag ng 'tl;dr:' sa dulo ng isang bahagi ng teksto. Pansinin kung paano nauunawaan ng modelo kung paano magsagawa ng maraming gawain nang walang karagdagang mga tagubilin. Maaari kang mag-eksperimento gamit ang mas mapanuring mga prompt kaysa sa tl;dr upang baguhin ang ugali ng modelo at iayon ang buod na iyong matatanggap(3).  

Kamakailang mga pag-aaral ay nagpakita ng malaking pag-unlad sa maraming gawain at benchmark sa NLP sa pamamagitan ng pre-training sa isang malaking korpus ng teksto kasunod ng fine-tuning sa isang tiyak na gawain. Bagama't kadalasan ay task-agnostic ang arkitektura, ang pamamaraang ito ay nangangailangan pa rin ng task-specific na fine-tuning datasets na may libo-libo o sampu-sampung libong mga halimbawa. Sa kabilang banda, ang mga tao ay kadalasang kayang magsagawa ng bagong gawain sa wika mula lamang sa ilang halimbawa o simpleng mga tagubilin - isang bagay na mahirap pa ring gawin ng mga kasalukuyang sistema ng NLP. Dito ipinapakita namin na ang pag-scale up ng mga language model ay malaki ang naitutulong sa task-agnostic, few-shot na performance, na minsan ay umaabot pa sa antas ng pagkakakumpitensya sa mga dating state-of-the-art na pamamaraan ng fine-tuning. 



Tl;dr

```python
prompt = "Recent work has demonstrated substantial gains on many NLP tasks and benchmarks by pre-training on a large corpus of text followed by fine-tuning on a specific task. While typically task-agnostic in architecture, this method still requires task-specific fine-tuning datasets of thousands or tens of thousands of examples. By contrast, humans can generally perform a new language task from only a few examples or from simple instructions - something that current NLP systems still largely struggle to do. Here we show that scaling up language models greatly improves task-agnostic, few-shot performance, sometimes even reaching competitiveness with prior state-of-the-art fine-tuning approaches.\n\nTl;dr"
```

## I-uri ang Teksto  
#### Hamon  
I-uri ang mga item sa mga kategoryang ibinigay sa panahon ng inferensya. Sa sumusunod na halimbawa, ibinibigay namin ang parehong mga kategorya at ang tekstong iuuri sa prompt(*playground_reference). 

Pagtatanong ng Customer: Hello, isa sa mga keys ng aking keyboard sa laptop ay nasira kamakailan at kailangan ko ng kapalit:

Naiuri na kategorya:


```python
prompt = "Classify the following inquiry into one of the following: categories: [Pricing, Hardware Support, Software Support]\n\ninquiry: Hello, one of the keys on my laptop keyboard broke recently and I'll need a replacement:\n\nClassified category:"
print(prompt)
```

```python
#Setting a few additional, typical parameters during API Call

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":prompt},],
  store=False,)

response.output_text
```
