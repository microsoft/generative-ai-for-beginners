### 2. Mag-import ng mga helper libraries at gumawa ng mga kredensyal


```python
import os
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("OPENAI_API_KEY","")
assert API_KEY, "ERROR: OpenAI Key is missing"

client = OpenAI(
    api_key=API_KEY
    )
```

### 3. Paghahanap ng tamang modelo  
Ang mga modelo tulad ng GPT-4o at GPT-4o mini ay nakakaintindi at nakakagawa ng natural na wika, at available sa OpenAI platform na may iba't ibang antas ng lakas at bilis na angkop para sa iba't ibang gawain.


```python
# Select a general purpose chat model
model = "gpt-4o-mini"
```

## 4. Disenyo ng Prompt  

"Ang mahika ng malalaking modelo ng wika ay sa pamamagitan ng pagsasanay upang mabawasan ang error sa prediksyon sa napakalaking dami ng teksto, natututuhan ng mga modelo ang mga konseptong kapaki-pakinabang para sa mga prediksiyong ito. Halimbawa, natututuhan nila ang mga konseptong tulad ng"(1):

#### Paano kontrolin ang isang malaking modelo ng wika  
"Sa lahat ng mga input sa isang malaking modelo ng wika, ang pinaka-maimpluwensiya ay ang prompt ng teksto(1).

Maaaring i-prompt ang malalaking modelo ng wika upang maglabas ng output sa ilang mga paraan:

Instruksiyon: Sabihin sa modelo kung ano ang gusto mo
Kumpletuhin: Hikayatin ang modelo na tapusin ang simula ng gusto mo
Demonstrasyon: Ipakita sa modelo kung ano ang gusto mo, sa alinman sa:
Ilang mga halimbawa sa prompt
Maraming daan o libong halimbawa sa isang fine-tuning na training dataset"



#### May tatlong pangunahing gabay sa paggawa ng mga prompt:

**Ipakita at sabihin**. Gawing malinaw kung ano ang gusto mo sa pamamagitan ng mga instruksiyon, mga halimbawa, o kumbinasyon ng dalawa. Kung gusto mong i-ranggo ng modelo ang isang listahan ng mga bagay ayon sa alpabetikal na ayos o i-klasipika ang isang talata batay sa damdamin, ipakita sa modelo na iyon ang gusto mo.

**Magbigay ng kalidad na data**. Kung sinusubukan mong bumuo ng isang classifier o gustong sundan ng modelo ang isang pattern, siguraduhing may sapat na mga halimbawa. Siguraduhing i-proofread ang iyong mga halimbawa — karaniwan nang matalino ang modelo upang makita ang mga simpleng kamalian sa pagbabaybay at magbigay ng tugon, ngunit maaari ring isipin nitong sinasadya ito at makaapekto sa tugon.

**Suriin ang iyong mga setting.** Kinokontrol ng mga setting na temperature at top_p kung gaano katiyak ang modelo sa pagbuo ng tugon. Kung humihingi ka ng tugon na may isang tama lang na sagot, gusto mong i-set ito sa mababa. Kung naghahanap ka ng mas sari-saring mga tugon, posibleng gusto mong itaas ito. Ang pinaka-madalas na pagkakamali ng mga tao sa mga setting na ito ay ang pag-aakalang ito ay mga kontrol ng "talino" o "kreatibidad."


Pinagmulan: https://learn.microsoft.com/azure/ai-services/openai/overview


### 5. Isumite!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```

## Ibuod ang Teksto  
#### Hamon  
Ibuod ang teksto sa pamamagitan ng pagdaragdag ng 'tl;dr:' sa dulo ng isang bahagi ng teksto. Pansinin kung paano naiintindihan ng modelo kung paano isagawa ang iba't ibang gawain nang walang karagdagang mga tagubilin. Maaari kang mag-eksperimento gamit ang mas detalyadong mga prompt kaysa sa tl;dr upang baguhin ang kilos ng modelo at i-customize ang buod na iyong matatanggap(3).  

Ipinakita ng mga kamakailang pag-aaral ang malalaking pag-unlad sa maraming gawain at benchmark sa NLP sa pamamagitan ng paunang pagsasanay sa isang malaking korpus ng teksto kasunod ng pag-fine-tune sa isang partikular na gawain. Habang karaniwang walang kaakibat na gawain sa estruktura, nangangailangan pa rin ang metodong ito ng mga dataset ng fine-tuning na partikular sa gawain na may libu-libo o sampu-sampung libong halimbawa. Sa kabilang banda, karaniwang nagagawa ng mga tao ang isang bagong gawain sa wika mula sa ilang mga halimbawa lamang o mula sa mga simpleng tagubilin - isang bagay na kasalukuyang lalong nahihirapan ang mga sistema ng NLP na gawin. Dito ipinapakita namin na ang pagpapalaki ng mga modelo ng wika ay lubos na nagpapabuti sa task-agnostic, few-shot na pagganap, kung minsan ay umaabot pa sa pagiging kompetitibo sa mga naunang nangungunang pamamaraan ng fine-tuning.  



Tl;dr
