### ការហៅមុខងារ ដើម 

Llama 3.1 ត្រូវបានបង្រួមបង្រួមចិត្តឲ្យមានប្រសិទ្ធភាពក្នុងការហៅមុខងារ ឬឧបករណ៍ជាងមុន។ វាក៏មានឧបករណ៍ពីរដែលស្វ័យប្រវត្តិ​ដែលគំរូអាចដឹងថាត្រូវប្រើដោយផ្អែកលើសំណើរពីអ្នកប្រើផងដែរ។ ឧបករណ៍ទាំងនេះមានជា៖ 

- **Brave Search** - អាចប្រើសម្រាប់ទទួលបានព័ត៌មានទាន់សម័យដូចជាការព្យាករណ៍អាកាសធាតុខណៈពេលធ្វើការស្វែងរកតាមវ៉េប 
- **Wolfram Alpha** - អាចប្រើសម្រាប់ការគណនាគណិតវិទ្យាដោយស្មុគស្មាញជាងនេះ ដូច្នេះ ការសរសេរកម្មវិធីផ្ទាល់ខ្លួនគឺមិនចាំបាច់ទេ។ 

អ្នកក៏អាចបង្កើតឧបករណ៍ផ្ទាល់ខ្លួនដែល LLM អាចហៅបានផងដែរ។ 

ក្នុងឧទាហរណ៍កូដខាងក្រោម៖ 

- យើងកំណត់ឧបករណ៍ដែលមានស្រាប់ (brave_search, wolfram_alpha) នៅក្នុងសំណើប្រព័ន្ធ។  
- ផ្ញើសំណើរពីអ្នកប្រើដែលសួរអំពីអាកាសធាតុលើទីក្រុងមួយ។  
- LLM នឹងឆ្លើយតបជាមួយការហៅឧបករណ៍ទៅឧបករណ៍ Brave Search ដែលនឹងមើលទៅដូចជា `<|python_tag|>brave_search.call(query="Stockholm weather")` 

*សម្គាល់៖ ឧទាហរណ៍នេះគ្រាន់តែធ្វើការហៅឧបករណ៍ប៉ុណ្ណោះ ប្រសិនបើអ្នកចង់ទទួលបានលទ្ធផល អ្នកនឹងត្រូវបង្កើតគណនីឥតគិតថ្លៃនៅលើទំព័រ Brave API ហើយកំណត់មុខងារផ្ទាល់ខ្លួន។*


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"
model_name = "meta-llama-3.1-405b-instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)



```

### ការគាំទ្រជាប្រភេទច្រើនជាមួយ Llama 3.2


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
