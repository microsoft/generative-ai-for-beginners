### бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶ Function Calling 

бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҺбһ¶бһҳбһҪбһҷбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһӮбҹ’бһӮбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһ”бҹ’бһҡбһ¶бһҖбһҠбһҗбһ¶бһўбҹ’бһ“бһҖбһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһўбҹҶбһҺбһ¶бһ…бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ’бҹҶбһ’бҹҶ (LLMs)бҹ” бһҹбһ„бҹ’бһғбһ№бһҳбһҗбһ¶бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһҖбҹҶбһҺбһҸбҹӢбһҒбҹ’бһӣбҹҮбҹ—бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһңбһ¶бһ•бһ„бһҠбҹӮбһҡбҹ” 

Function Calling бһӮбһәбһҮбһ¶бһҖбһ¶бһҡбһ…бҹҶбһ“бһ»бһ…бһҳбһҪбһҷбһ“бҹғбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ Azure Open AI бһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ„бһҸбһҖбҹҶбһҺбһҸбҹӢбһҒбҹ’бһӣбҹҮбҹ—бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–  
1) бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһ…бһҳбҹ’бһӣбһҫбһҷбһҳбһ¶бһ“бһ—бһ¶бһ–бһҹбһҳбһҡбһҳбҹ’бһҷ 
2) бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ”бҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бһёбһ”бҹ’бһҡбһ—бһ–бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҖбҹ’бһ“бһ»бһ„бһ”бһҡбһ·бһ”бһ‘бһҹбһ“бҹ’бһ‘бһ“бһ¶ 

бһҳбһ»бһ“бһ–бҹҒбһӣбһҳбһ¶бһ“ function calling бһ…бһҳбҹ’бһӣбһҫбһҷбһ–бһё LLM бһҸбҹӮбһ„бһҸбҹӮбһўбһ“бһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҳбһ·бһ“бһҳбһ¶бһ“бһҹбһҳбһҡбһҳбҹ’бһҷбҹ” бһўбҹ’бһ“бһҖбһўwickeys бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ‘бһ¶бһҳбһ‘бһ¶бһҡбһІбҹ’бһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҠбҹҸбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһҠбһҫбһҳбҹ’бһ”бһёбһ’бһ¶бһ“бһ¶бһҗбһ¶бһ–бһҪбһҖбһӮбҹҒбһ…бһ”бҹӢбһҠбҹҶбһҺбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһ¶бһ“бһ…бҹҶбһ–бҹ„бҹҮбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһ“бһёбһҳбһҪбһҷбҹ—бһ“бҹғбһ…бһҳбҹ’бһӣбһҫбһҷбҹ”

```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```

бһҷбһҫбһ„бһ…бһ„бҹӢбһ•бҹ’бһүбһҫбһҡбһ“бҹҒбҹҮбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ”бҹӮбһҖбһ…бҹҶбһҺбһ»бһ…бһҠбҹӮбһӣбһҹбҹҶбһҒбһ¶бһ“бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” бһҠбһјбһ…бһ“бҹҒбҹҮ бһҷбһҫбһ„бһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһ–бһёбһҡбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһҺбҹӮбһ“бһ¶бҹҶ LLMбҹ–


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ“бҹҒбҹҮбһҡбһҪбһ… бһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһ–бһҪбһҖбһңбһ¶бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `openai.ChatCompletion`бҹ” бһҷбһҫбһ„бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһҹбҹҶбһҺбһҫбһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡ `messages` бһ бһҫбһҷбһ…бһ¶бһҸбҹӢбһҸбһ¶бҹҶбһ„бһҸбһҪбһ“бһ¶бһ‘бһёбһ‘бҹ…бһҮбһ¶ `user`бҹ” бһ“бҹҒбҹҮбһӮбһәбһҠбһҫбһҳбҹ’бһ”бһёбһ…бһҳбҹ’бһӣбһ„бһҹбһ¶бһҡбһҳбһҪбһҷбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҹбһҡбһҹбҹҒбһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢ chatbotбҹ”


```python
import os
import json
from openai import AzureOpenAI
from dotenv import load_dotenv
load_dotenv()

client = AzureOpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  # this is also the default, it can be omitted
  api_version = "2023-07-01-preview"
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһўбһ¶бһ…бһ•бҹ’бһүбһҫбһҹбҹҶбһҺбһҫбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бҹ”


```python
openai_response1 = client.chat.completions.create(
 model=deployment,    
 messages = [{'role': 'user', 'content': prompt1}]
)
openai_response1.choices[0].message.content 
```

```python
openai_response2 = client.chat.completions.create(
 model=deployment,    
 messages = [{'role': 'user', 'content': prompt2}]
)
openai_response2.choices[0].message.content
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.choices[0].message.content)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.choices[0].message.content )
json_response2
```

### бһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ

**бһҖбһ¶бһҡбһ бҹ…бһ§бһ”бһҖбһҡбһҺбҹҚбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ…**  
Chatbots бһҳбһ¶бһ“бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһӣбҹ’бһўбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ•бҹ’бһҠбһӣбҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбһӣбҹӢбһҹбҹҶбһҺбһҪбһҡбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбҹ” бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ chatbots бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһ¶бһҡвҖӢбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһ”бһүбҹ’бһ бһ¶бһ•бҹ’бһҹбҹҒбһ„бҹ—бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ, бһҹбһ·бһҹбҹ’бһҹбһўбһ¶бһ…бһҹбҹ’бһ“бһҫбһІбҹ’бһҷ chatbot "бһҹбһјбһҳбһ•бҹ’бһүбһҫбһўбҹҠбһёбһҳбҹӮбһӣбһ‘бҹ…бһӮбҹ’бһҡбһјбһҡбһ”бһҹбҹӢбһҒбҹ’бһүбһ»бҹҶбһҗбһ¶бһҒбҹ’бһүбһ»бҹҶбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҮбҹҶбһ“бһҪбһҷбһ”бһ“бҹ’бһҗбҹӮбһҳбһҹбҹ’бһҸбһёбһ–бһёбһҳбһ»бһҒбһңбһ·бһҮбҹ’бһҮбһ¶бһ“бҹҒбҹҮ"бҹ” бһ“бҹҒбҹҮбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһҠбҹ„бһҷбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ `send_email(to: string, body: string)` 

**бһ”бһ„бҹ’бһҖбһҫбһҸ API бһ¬бһҹбҹҶбһҺбһҪбһҡбһ‘бҹ…бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷ**  
бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһўбһ¶бһ…бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҳбҹ’бһӣбҹӮбһ„бһ‘бҹ…бһҮбһ¶бһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“ бһ¬ API requestбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһӮбҹ’бһҡбһјбһўбһ¶бһ…бһҹбһҪбһҡ "бһ“бһҡбһҺбһ¶бһҒбҹ’бһӣбҹҮбһҮбһ¶бһҹбһ·бһҹбҹ’бһҹбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһҖбһ¶бһҡбһ„бһ¶бһҡбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ”бһ“бҹ’бһҹбҹҶбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ" бһҠбҹӮбһӣбһўбһ¶бһ…бһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ `get_completed(student_name: string, assignment: int, current_status: string)`

**бһ”бһ„бҹ’бһҖбһҫбһҸбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһ¶бһ“бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’**  
бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһўбһ¶бһ…бһҷбһҖбһўбһҸбҹ’бһҗбһ”бһ‘бһҳбһҪбһҷбһ…бҹҶбһҺбһ»бһ… бһ¬ CSV бһ бһҫбһҷбһ”бҹ’бһҡбһҫ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһҷбһҖбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҹбҹҶбһҒбһ¶бһ“бҹӢбһ…бҹҒбһүбһ–бһёбһңбһ¶бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһҹбһ·бһҹбҹ’бһҹбһўбһ¶бһ…бһ”бһҳбҹ’бһӣбҹӮбһ„бһўбһҸбҹ’бһҗбһ”бһ‘ Wikipedia бһўбҹҶбһ–бһёбһҖбһ·бһ…бҹ’бһ…бһ–бҹ’бһҡбһҳбһ–бҹ’бһҡбҹҖбһ„бһ“бһјбһңбһҹбһ“бҹ’бһҸбһ·бһ—бһ¶бһ– бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҸбһҡбҹҶбһӣбһ№бһҖ AIбҹ” бһ“бҹҒбҹҮбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡ `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)`

`role` бһўбһ¶бһ…бһҮбһ¶бһўбҹ’бһңбһёбһҳбҹ’бһҷбҹүбһ¶бһ„бһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳ `system` (бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бҹ’бһ”бһ¶бһ”бҹӢ), `assistant` (бһҳбҹүбһјбһҠбҹӮбһӣ) бһ¬ `user` (бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ…бһ»бһ„бһ”бһүбҹ’бһҮбһ¶)бҹ” бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһңбһ¶бһҮбһ¶ `user` бһ“бһ·бһ„бһҹбҹҶбһҺбһҪбһҡбһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҳбһҪбһҷбҹ”

### бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһ“бһ»бһӮбһҳбһ“бҹҚбҹ” 

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҳбһҪбһҷ бһ“бһ·бһ„бһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡбһ“бҹғбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһ“бҹ„бҹҮбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҸбҹӮбһҳбһҪбһҷбһӮбһҸбҹӢбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮ `search_courses` бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһ“бһ№бһ„бһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ”бһ¶бһ“бҹ” 

**бһҹбҹҶбһҒбһ¶бһ“бҹӢ**бҹ– бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҹбһ¶бһҡвҖӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһ“бһҪбһ“ token бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбһҫбҹ”


```python
functions = [
   {
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**бһ“бһ·бһҷбһҳбһ“бҹҗбһҷ**

`name` - бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһҠбҹӮбһӣвҖӢбһҷбһҫбһ„вҖӢбһ…бһ„бҹӢвҖӢбһІбҹ’бһҷвҖӢбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹ„бҹҮвҖӢбһҸбҹ’бһҡбһјбһңвҖӢбһ”бһ¶бһ“бһ бҹ…бҹ”

`description` - бһ“бҹҒбҹҮвҖӢбһӮбһәвҖӢбһҮбһ¶вҖӢбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶вҖӢбһўбҹҶбһ–бһёвҖӢбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣвҖӢбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹҒбҹҮвҖӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһңбһ¶вҖӢбһҹбҹҶбһҒбһ¶бһ“бҹӢвҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһҖбһ¶бһҡвҖӢбһ’бҹ’бһңбһҫвҖӢбһІбҹ’бһҷвҖӢбһҳбһ¶бһ“вҖӢбһ—бһ¶бһ–бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ бһ“бһ·бһ„вҖӢбһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢбҹ”

`parameters` - бһ”бһүбҹ’бһҮбһёвҖӢбһҸбһҳбҹ’бһӣбҹғ бһ“бһ·бһ„вҖӢбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷвҖӢбһҠбҹӮбһӣвҖӢбһўбҹ’бһ“бһҖвҖӢбһ…бһ„бҹӢбһІбҹ’бһҷвҖӢбһҳбҹүбһјбһҠбҹӮбһӣвҖӢбһ•бһӣбһ·бһҸвҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһҖбһ¶бһҡвҖӢбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷвҖӢбһҠбҹӮбһӣвҖӢбһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһ“бһ¶бһ“бһ¶вҖӢбһ“бһ№бһ„вҖӢбһҸбҹ’бһҡбһјбһңвҖӢбһҠбһ¶бһҖбҹӢвҖӢбһ”бһ„бҹ’бһ бҹ„бҹҮвҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһңбһ¶бҹ”

`properties` - бһ”бһүбҹ’бһҮбһёвҖӢбһҸбһҳбҹ’бһӣбҹғвҖӢбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢвҖӢбһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣвҖӢбһ“бһ№бһ„вҖӢбһ”бҹ’бһҡбһҫвҖӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢвҖӢбһҖбһ¶бһҡвҖӢбһҶбҹ’бһӣбһҫбһҷбһҸбһ”вҖӢбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

`name` - бһҲбҹ’бһҳбҹ„бҹҮвҖӢбһ“бҹғвҖӢбһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣвҖӢбһ“бһ№бһ„вҖӢбһ”бҹ’бһҡбһҫвҖӢбһҖбҹ’бһ“бһ»бһ„вҖӢбһҖбһ¶бһҡвҖӢбһҶбҹ’бһӣбһҫбһҷвҖӢбһҸбһ”вҖӢбһҠбҹӮбһӣвҖӢбһ”бһ¶бһ“вҖӢбһ”бҹӮбһ”бһ”бһ‘бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷвҖӢбһҡбһ”бһҹбҹӢвҖӢбһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһ“бҹҒбҹҮбҹ”

`description` - бһ–бһ·бһ–бһҺбҹҢбһ“бһ¶вҖӢбһўбҹҶбһ–бһёвҖӢбһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢвҖӢбһ“бҹҒбҹҮбҹ”

**бһҮбһ¶бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢ**

`required` - бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲвҖӢбһ”бһ¶бһ…бҹӢвҖӢбһҳбһ·бһ“бһўбһ¶бһ…бһҒбҹ’бһңбҹҮвҖӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢвҖӢбһҖбһ¶бһҡвҖӢбһ бҹ…вҖӢбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһІбҹ’бһҷвҖӢбһ”бһүбҹ’бһ…бһ”бҹӢбҹ”


### бһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ
бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһҹбһ–бҹ’бһңбһҗбҹ’бһ„бҹғбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ бҹ… API Chat Completionбҹ” бһҷбһҫбһ„бһ’бҹ’бһңбһҫбһ“бҹҒбҹҮбһҠбҹ„бһҷбһ”бһ“бҹ’бһҗбҹӮбһҳ `functions` бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбҹ” бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ `functions=functions`бҹ”

бһҳбҹӮбһ“бһ‘бһ¶бҹҶбһ„бһҳбһ¶бһ“бһҮбһҳбҹ’бһҡбһҫбһҹбһҳбһҪбһҷбһҠбһҫбһҳбҹ’бһ”бһёбһҖбҹҶбһҺбһҸбҹӢ `function_call` бһ‘бҹ…бһҮбһ¶ `auto`бҹ” бһ“бҹҒбҹҮбһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһҷбһҫбһ„бһ“бһ№бһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷ LLM бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҳбһ»бһҒбһ„бһ¶бһҡбһҺбһ¶бһҳбһҪбһҷбһҠбҹӮбһӣбһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ… dб»ұa trГӘnбһҹбһ¶бһҡвҖӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһҡбһҪбһ…бһ бҹ…бһңбһ¶ бһҮбҹҶбһ“бһҪбһҹбһҖбһ¶бһҡбһ‘бһ‘бһҪбһӣбһҒбһ»бһҹбһҸбҹ’бһҡбһјбһңбһҠбҹ„бһҷбһҒбҹ’бһӣбһҪбһ“бһҜбһ„бҹ”


```python
response = client.chat.completions.create(model=deployment, 
                                        messages=messages,
                                        functions=functions, 
                                        function_call="auto") 

print(response.choices[0].message)
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһҠбҹӮбһӣбһ“бһ№бһ„бһ бҹ… API бһҡбһ”бһҹбҹӢ Microsoft Learn бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһёбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бҹ–


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

бһҮбһ¶бһ‘бһҳбҹ’бһӣбһ¶бһ”бҹӢбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸ бһҷбһҫбһ„бһ“бһ№бһ„бһҳбһҫбһӣбһҗбһ¶ бһҸбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ…бһ„бҹӢбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһ¬бһ‘бҹҒбҹ” бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹ„бҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“ бһ бһҫбһҷбһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ…бҹ”
бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһҷбһҫбһ„бһ“бһ№бһ„бһҷбһҖбһўбһ¶бһӮбһ»бһҷбһҳбҹӮбһ“ (arguments) бһҡбһ”бһҹбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһ бһҫбһҷбһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһўбһ¶бһӮбһ»бһҷбһҳбҹӮбһ“бһ–бһё LLMбҹ”

бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ бһҷбһҫбһ„бһ“бһ№бһ„бһ—бҹ’бһҮбһ¶бһ”бҹӢбһҹбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһ“бһ·бһ„бһҸбһҳбҹ’бһӣбҹғбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбһҠбҹ„бһҷбһҹбһ¶бһҡбһҡбһ”бһҹбҹӢ `search_courses`бҹ” бһ“бҹҒбҹҮбһ•бҹ’бһҸбһӣбҹӢбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҠбҹӮбһӣ LLM бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ”бҹ’бһҡбһ–бҹғбһҺбһёбҹ”


```python
# Check if the model wants to call a function
if response_message.function_call.name:
    print("Recommended Function call:")
    print(response_message.function_call.name)
    print()

    # Call the function. 
    function_name = response_message.function_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(response_message.function_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the assistant response and function response to the messages
    messages.append( # adding assistant response to messages
        {
            "role": response_message.role,
            "function_call": {
                "name": function_name,
                "arguments": response_message.function_call.arguments,
            },
            "content": None
        }
    )
    messages.append( # adding function response to messages
        {
            "role": "function",
            "name": function_name,
            "content":function_response,
        }
    )

```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһҹбһ¶бһҡбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ“бҹ’бһҗбҹӮбһҳбһ…бһјбһӣбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһҷбһҫбһ„бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ…бһҳбҹ’бһӣбһҫбһҷбһҮбһ¶бһ—бһ¶бһҹбһ¶бһ‘бҹҶбһ“бһҫбһ”бһҳбһҪбһҷ бһҮбҹҶбһ“бһҪбһҹбһІбҹ’бһҷбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷ JSON APIбҹ”
