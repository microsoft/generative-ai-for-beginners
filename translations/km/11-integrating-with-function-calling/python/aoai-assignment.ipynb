## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ

бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбһ“бһ№бһ„бһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’:

- бһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ Azure OpenAI 
- бһҡбһ”бҹҖбһ”бһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷ

## бһ‘бһҹбҹ’бһҹбһ“бҹҲбһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶

- бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ
- бһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһ…бҹҶбһҺбһ»бһ…бһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ Function Call бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Azure Open AI Service
- бһҖбһ¶бһҡбһҡбһ…бһ“бһ¶бһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣ: 
 - `Azure Open AI` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҮбһҮбҹӮбһҖбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ  
 - `Microsoft Learn Catalog API` бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҡбһҖбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҹбҹҶбһҺбһҫбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ  
 - `Function Calling` бһҠбһҫбһҳбҹ’бһ”бһёбһҷбһҖбһҹбҹҶбһҺбһҪбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһ бһҫбһҷбһ•бҹ’бһүбһҫбһңбһ¶бһ‘бҹ…бһ бҹ’бһңаёұаёҮаёҒа№Ңбһҹбһ·бһ“бһҳбһҪбһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶ APIбҹ”

### бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶ Function Calling  

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһҺбһ¶бһҳбһҪбһҷбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһӮбҹ’бһӮбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһ’бҹ’бһӣбһ¶бһ”бҹӢбһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһўбҹҶбһҺбһ¶бһ…бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ’бҹҶбҹ— (LLMs)бҹ” бһҹбһ„бҹ’бһғбһ№бһҳбһҗбһ¶ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһңбһ¶бһ•бһ„бһҠбҹӮбһҡбҹ”  

Function Calling бһӮбһәбһҮбһ¶бһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһ“бҹғбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ Azure Open AI бһҠбһҫбһҳбҹ’бһ”бһёбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹҲ  
1) бһҡбһјбһ”а№Ғаёҡаёҡбһ…бһҳбҹ’бһӣбһҫбһҷбһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶  
2) бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ”бҹ’бһҡбһҫбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бһёбһ”бҹ’бһҡбһ—бһ–бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бһҡбһ·бһ”бһ‘бһҮбһҮбҹӮбһҖ  

бһҳбһ»бһ“бһ“бһ№бһ„бһ”бҹ’бһҡбһҫ function calling бһ…бһҳбҹ’бһӣбһҫбһҷбһ–бһё LLM бһҳбһ·бһ“бһҳбһ¶бһ“бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’ бһ“бһ·бһ„бһҳбһ·бһ“бһҹбҹ’бһҳбһҫбһӮбҹ’бһ“бһ¶бһ‘бҹҒбҹ” бһўбҹ’бһ“бһҖбһўбһ—бһ·бһңбһҢбҹ’бһҚбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһӮбҹҒбһ…бһ¶бһҸбҹӢбһ‘бһ»бһҖбһІбҹ’бһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһү бһҠбһҫбһҳбҹ’бһ”бһёбһ’бһ¶бһ“бһ¶бһҗбһ¶бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ¶бһӣбҹӢбһ”бһҳбҹ’бһӣбҹӮбһ„бһҳбһҪбһҷбҹ—бһ”бһ¶бһ“бҹ”


бһҷбһҫбһ„бһ…бһ„бҹӢбһ•бҹ’бһүбһҫбһҡбһ“бҹҒбҹҮбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһҖбһҹбҹ’бһҡбһ¶бһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ” бһ–бҹҒбһӣбһҖбҹ’бһҡбҹ„бһҷбһ“бҹҒбҹҮбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҠбһҫбһҳбҹ’бһ”бһёбһ•бҹ’бһүбһҫбһҡвҖӢбһ‘бҹ… API бһ¬бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

бһҷбһҫбһ„бһ…бһ„бҹӢбһ•бҹ’бһүбһҫбһҡбһӣбһ»бһҷбһ“бҹҒбҹҮбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ”бҹӮбһҖбһ•бҹ’бһ“бҹӮбһҖбһҠбҹӮбһӣбһҹбҹ’бһҗбһ·бһҸбһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ…бҹҶбһ–бҹ„бҹҮбһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҷбһҫбһ„бһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһ“бһҪбһҡбһ–бһёбһҡбһҠбһјбһ…бһӮбҹ’бһ“бһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһҺбҹӮбһ“бһ¶бҹҶ LLMбҹ– 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘бһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ“бҹҒбҹҮбһҡбһҪбһ… бһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһ–бһҪбһҖбһңбһ¶бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `client.responses.create`бҹ” бһҷбһҫбһ„бһ•бҹ’бһ‘бһ»бһҖбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘бһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡ `input` бһ бһҫбһҷбһ…бһ¶бһҸбҹӢбһҸбһ¶бҹҶбһ„бһҸбһҪбһ“бһ¶бһ‘бһёбһ‘бҹ…бһҮбһ¶ `user`бҹ” бһ“бҹҒбҹҮбһӮбһәбһҠбһҫбһҳбҹ’бһ”бһёбһ…бһҳбҹ’бһӣбһ„бһҹбһ¶бһҡбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҹбһҡбһҹбҹҒбһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢ chatbotбҹ”



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһўбһ¶бһ…бһ•бҹ’бһүбһҫбһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бҹ”


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```

### бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбҹ”

**бһҹбҹҶбһҒбһ¶бһ“бҹӢ**бҹ– бһҳбһ»бһҒбһ„бһ¶бһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбһ¶бһҡ бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһ“бһҪбһ“бһ’бһ»бһ„бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбҹ’бһҡбһ¶бһ”бҹӢбҹ”

**бһ“бһ·бһҷбһҳбһ“бҹҗбһҷ** 

`name` - бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһІбҹ’бһҷбһ бҹ…бҹ” 

`description` - бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһҖбһ¶бһҡбһ–бһҺбҹҢбһ“бһ¶бһўбҹҶбһ–бһёбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһҳбһ»бһҒбһ„бһ¶бһҡбһҸбҹ’бһҡбһјбһңбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһңбһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһ“бһ·бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ 

`parameters` - бһ”бһүбҹ’бһҮбһёбһҸбһҳбҹ’бһӣбҹғ бһ“бһ·бһ„бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһІбҹ’бһҷбһҳбҹүбһјбһҢбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹ’бһ“бһ»бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһңбһ¶ 


`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ•бҹ’бһ‘бһ»бһҖ 

`properties` - бһ”бһүбҹ’бһҮбһёбһҖбһҸбҹ’бһҸбһ¶бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҠбҹӮбһӣбһҳбҹүбһјбһҢбҹӮбһӣбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһңбһ¶ 


`name` - бһҲбҹ’бһҳбҹ„бҹҮбһҖбһҸбҹ’бһҸбһ¶бһҠбҹӮбһӣбһҳбҹүбһјбһҢбҹӮбһӣбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҡбһ”бһҹбҹӢбһңбһ¶ 

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғбһҖбһҸбҹ’бһҸбһ¶бһ“бҹҒбҹҮ 

`description` - бһҖбһ¶бһҡбһ–бһҺбҹҢбһ“бһ¶бһўбҹҶбһ–бһёбһҖбһҸбҹ’бһҸбһ¶бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ 


**бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹ**

`required` - бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһұбҹ’бһҷбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢ 


### бһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ
бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһҘбһЎбһјбһңбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶вҖӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ бҹ…бһ‘бҹ…бһҖбһ¶бһ“бҹӢ Chat Completion APIбҹ” бһҷбһҫбһ„бһ’бҹ’бһңбһҫбһҠбһјбһ…бһ“бҹҒбҹҮбһҠбҹ„бһҷбһ”бһ“бҹ’бһҗбҹӮбһҳ `functions` бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹҶбһҺбһҫбҹ” бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ `functions=functions`бҹ”

бһҖбҹҸбһҳбһ¶бһ“бһҮбһҳбҹ’бһҡбһҫбһҹбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбҹҶбһҺбһҸбҹӢ `function_call` бһ‘бҹ…бһҮбһ¶ `auto` бһ•бһ„бһҠбҹӮбһҡбҹ” бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһҷбһҫбһ„бһ“бһ№бһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷ LLM бһҹбһҳбҹ’бһҡбҹҒбһ…бһҗбһ¶бһҸбһҫбһҳбһ»бһҒбһ„бһ¶бһҡбһҺбһ¶бһҳбһҪбһҷбһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ… dб»ұa trГӘnбһҹбһ¶бһҡвҖӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ бһҮбҹҶбһ“бһҪбһҹбһҖбһ¶бһҡбһ…бҹӮбһ„бһ”бһүбҹ’бһҮбһ¶вҖӢбһҠбҹ„бһҷбһҒбҹ’бһӣбһҪбһ“бһҜбһ„бҹ”


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

## 3. бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷбҹ”


бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ–бһҪбһҖбһҷбһҫбһ„бһ”бһ¶бһ“бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбһ…бһ“бһ¶бһ‘бһҳбҹ’бһҡбһ„бҹӢбһ…бҹҒбһүбһ–бһё LLM бһҖбҹ’бһҡбҹ„бһҷбһҳбһҖбһҘбһЎбһјбһңбһ“бҹҒбҹҮбһ–бһҪбһҖбһҷбһҫбһ„бһўбһ¶бһ…бһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷбһ”бһ¶бһ“бһ бһҫбһҷбҹ”

### бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ

бһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһҷбһҫбһ„ бһӮбҹҒбһҸбҹ’бһҡбһјбһңбһ’бҹ’бһңбһҫбһҮбҹҶбһ бһ¶бһ“бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

бһҮбһ¶бһ„бһҠбҹҶбһ”бһјбһ„ бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ бҹ…бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ Open AI бһ бһҫбһҷбһҡбһҖбҹ’бһҹбһ¶бһҹбһ¶бһҡбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡбһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮбһҗбһ¶ `response_message`бҹ”


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹӮбһӣбһ“бһ№бһ„бһ бҹ… API Microsoft Learn бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһёбһҳбҹҒбһҡбҹҖбһ“: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢвҖӢбһ–бһёвҖӢбһ“бҹ„бҹҮ бһҷбһҫбһ„вҖӢбһ“бһ№бһ„вҖӢбһҷбһҖвҖӢбһў Р°СҖРіСғРјРөРҪСӮСӢ бһ“бҹғвҖӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһ бһҫбһҷвҖӢбһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„вҖӢбһ–бһҪбһҖбһңбһ¶вҖӢбһ‘бҹ…вҖӢбһ“бһ№бһ„вҖӢбһў Р°СҖРіСғРјРөРҪСӮСӢвҖӢбһ–бһё LLMбҹ”

бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ бһҷбһҫбһ„вҖӢбһ“бһ№бһ„вҖӢбһ”бһ“бҹ’бһҗбҹӮбһҳвҖӢбһҹбһ¶бһҡвҖӢбһ бҹ…вҖӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһ“бһ·бһ„вҖӢбһҸбһҳбҹ’бһӣбҹғвҖӢбһҠбҹӮбһӣвҖӢбһҸбҹ’бһҡбһјбһңвҖӢбһ”бһ¶бһ“вҖӢбһҸбҹ’бһҡбһӣбһ”бҹӢвҖӢбһҳбһҖвҖӢбһ–бһёвҖӢбһҹбһ¶бһҡ `search_courses`бҹ” бһ“бҹҒбҹҮвҖӢбһ•бҹ’бһҠбһӣбҹӢвҖӢбһұбҹ’бһҷ LLM бһҮбһ¶бһҳбһҪбһҷвҖӢбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“вҖӢбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢвҖӢбһҠбҹӮбһӣвҖӢбһңбһ¶вҖӢбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһё
бһҸбһ”вҖӢбһҹбҹ’бһ“бһ„вҖӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢвҖӢбһҠбҹ„бһҷвҖӢбһ”бҹ’бһҡбһҫвҖӢбһ—бһ¶бһҹбһ¶вҖӢбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бҹ” 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```
