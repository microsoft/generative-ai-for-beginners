## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ  

бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбһ“бһ№бһ„бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҸбһ”бҹӢбҹ–  
- бһҸбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҮбһ¶бһўбҹ’бһңбһё бһ“бһ·бһ„бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһңбһ¶бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…  
- бһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ OpenAI  
- бһҡбһ”бҹҖбһ”бһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷ  

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶  

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҠбһ№бһ„бһ–бһёбһҡбһ”бҹҖбһ”бһ“бһ·бһ„бһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбҹ–  

- бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ  
- бһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI  
- бһҡбһ…бһ“бһ¶бһҡбһ”бҹҖбһ”бһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ  


## бһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ 

бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ…бһ„бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ бҹ’бһӮбҹ„бһӣбһўбһ”бҹӢбһҡбҹҶбһҡбһ”бһҹбҹӢбһҷбһҫбһ„ бһҠбҹӮбһӣбһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбһ…бһ·бһҸбҹ’бһҸбһ”бһјбһҸбһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһӮбҹ’бһӮбһ”бһ…бҹ’бһ…бҹҒбһҖбһ‘бҹҒбһҹбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһҺбҹӮбһ“бһ¶бҹҶбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһҠбҹӮбһӣбһҹбһҳбһҡбһҳбҹ’бһҷбһ“бһ№бһ„бһҖбһҳбҹ’бһҡбһ·бһҸбһҮбҹҶбһ“бһ¶бһү бһҡбһ”бҹҖбһ”бһҸбҹҶбһҺбҹӮбһ„бһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“ бһ“бһ·бһ„бһ”бһ…бҹ’бһ…бҹҒбһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһҠбҹӮбһӣбһ…бһ¶бһ”бҹӢбһўбһ¶бһҡбһҳбҹ’бһҳбһҺбҹҚбҹ” 

бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҡбһңбһ¶бһ„ бҹ– 
 - `OpenAI` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҮбһҮбҹӮбһҖбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ 
 - `Microsoft Learn Catalog API` бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһҸбһ¶бһҳбһҹбҹҶбһҺбһҫбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ 
 - `Function Calling` бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһҹбҹҶбһҺбһҪбһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһ бһҫбһҷбһ•бҹ’бһүбһҫбһңбһ¶бһ‘бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҹбҹҶбһҺбһҫ API бҹ” 

бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ бһҳбһҖбһҳбһҫбһӣбһҳбһјбһӣбһ бҹҒбһҸбһ»бһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҮбһ¶бһ‘бһёбһҳбһҪбһҷ бҹ– 


### бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶ Function Calling  

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһҺбһ¶бһҳбһҪбһҷбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһҖбҹ’бһ“бһ»бһ„бһңбһӮбҹ’бһӮбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһ”бһ¶бһ“бһҷбһӣбҹӢбһ–бһёбһўбҹҶбһҺбһ¶бһ…бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ’бҹҶ (LLMs)бҹ” бһҹбһ„бҹ’бһғбһ№бһҳбһҗбһ¶бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһғбһҫбһүбһҖбҹҸбһҠбһјбһ…бһҮбһ¶бһ”бһүбҹ’бһ бһ¶бһӮбҹ’бһҡбһ”бҹӢбһҒбҹ’бһӣбҹҮбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһңбһ¶бҹ” 

Function Calling бһӮбһәбһҮбһ¶бһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹбһҳбһҪбһҷбһҡбһ”бһҹбҹӢбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҡбһ…бһ“бһ¶бһЎбһҫбһ„бһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ…бҹҶбһӣбһҫбһҷбһҠбҹӮбһӣбһҳбһ·бһ“бһҳбһ¶бһ“бһ—бһ¶бһ–бһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶:
- бһҳбһ»бһ“бһ–бҹҒбһӣбһҳбһ¶бһ“ function calling бһ…бҹҶбһӣбһҫбһҷбһ–бһёбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһ’бҹҶбһӮбһәбһӮбҹ’бһҳбһ¶бһ“бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’ бһ“бһ·бһ„бһҳбһ·бһ“бһҹбҹ’бһҡбһ”бһӮбҹ’бһ“бһ¶бҹ” бһўбҹ’бһ“бһҖбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ“бҹҚбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҸбҹҶбһҡбһјбһңбһұбҹ’бһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҖбһ¶бһҡбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбһјбһӣбһ…бҹҒбһүбһ“бһёбһҳбһҪбһҷбҹ—бҹ”

бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҮбһ¶бһҳбһҪбһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ…:
- бһҳбһ»бһ“бһ–бҹҒбһӣбһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹбһ“бҹҒбҹҮ бһңбһ¶бһҖбһәбһ–бһ·бһ”бһ¶бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ–бһёбһ•бҹ’бһ“бҹӮбһҖбһ•бҹ’бһҹбҹҒбһ„бҹ—бһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бһҡбһ·бһ”бһ‘бһҮбһҮбҹӮбһҖбҹ”

бһҠбҹ„бһҷбһҖбһ¶бһҡбһҸбһҳбҹ’бһӣбһҫбһ„бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ…бҹҶбһӣбһҫбһҷбһұбҹ’бһҷбһҳбһ¶бһ“бһҹбҹ’бһҸбһ„бҹӢбһҠбһ¶бһҡ бһ“бһ·бһ„бһ”бһҫбһҖбһўбһ“бһ»бһ”бһ¶бһҸбһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҳбһҪбһҷбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ…бһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһҡбһӣбһјбһ“ Function Calling бһ’бҹ’бһңбһҫбһұбҹ’бһҷбһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһҖбһ¶бһ“бҹӢбһҸбҹӮбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣ бһ“бһ·бһ„бһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһҖбһ¶бһҡбһҸбҹҶбһҡбһјбһңбһҖбһ¶бһҡбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбҹӮбһҳбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ–бһ¶бһҖбҹ’бһҷбһҹбһ»бҹҶбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ“бҹҒбҹҮбһҡбһҪбһ…бһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһ–бһҪбһҖбһңбһ¶бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ `client.responses.create` бҹ” бһҷбһҫбһ„бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһ–бһ¶бһҖбҹ’бһҷбһҹбһ»бҹҶбһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡ `input` бһ бһҫбһҷбһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„бһҸбһҪбһ“бһ¶бһ‘бһёбһ‘бҹ… `user` бҹ” бһ“бҹҒбҹҮбһӮбһәбһҠбһҫбһҳбҹ’бһ”бһёбһ–бһ·бһҹбҹ„бһ’бһҹбһ¶бһҡбһҳбһҪбһҷбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҠбҹӮбһӣбһҖбҹҶбһ–бһ»бһ„бһҸбҹӮбһ”бһүбҹ’бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„ chatbot бҹ”



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһўбһ¶бһ…бһ•бҹ’бһүбһҫбһҹбҹҶбһҺбһјбһҳбһ–бһҡбһ‘бһ¶бҹҶбһ„бһ–бһёбһҡбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бҹ” 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

бһҮбҹҶбһ бһ¶бһ“бһҠбҹҶбһ”бһјбһ„бһӮбһәбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһ¶бһҡвҖӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбҹ” бһ“бҹҒбҹҮбһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“вҖӢбһ”бҹӮбһ„бһ…бҹӮбһҖбһҮбһ¶бһҹбһҖбһҳбҹ’бһҳбһҠбҹ„бһҷбһҷбһҖбһҸбһҳбҹ’бһӣбҹғбһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһўбһҖбҹ’бһҹбһҡ бһ¬бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹӮбһ„бһ…бҹӮбһҖбһҸбһҳбҹ’бһӣбҹғбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһ“бҹҒбҹҮбһҮбһ¶бһӣбһҫбһҖбһ‘бһёбһҳбһҪбһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ’бҹ’бһңбһҫбһҖбһ¶бһҡвҖӢбһҮбһ¶бһҳбһҪбһҷ API Chat Completions бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңвҖӢбһҖбҹҶбһҺбһҸбҹӢ `role` бһ“бһ·бһ„ `content` бһ“бҹғбһҹбһ¶бһҡбҹ”  

бһӣбҹҒбһҒ `role` бһўбһ¶бһ…бһҮбһ¶ `system` (бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бҹ’бһ”бһ¶бһ”бҹӢ), `assistant` (бһҳбҹүбһјбһҠбҹӮбһӣ) бһ¬ `user` (бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ)бҹ” бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ, бһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһңбһ¶бһҮбһ¶ `user` бһ“бһ·бһ„бһҹбҹҶбһҺбһҪбһҡбһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҳбһҪбһҷбҹ”  


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷ бһ“бһ·бһ„бһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡвҖӢбһҡбһ”бһҹбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹ„бҹҮбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҳбһ»бһҒбһ„бһ¶бһҡбһҸбҹӮбһҳбһҪбһҷбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ `search_courses` бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ”бһ¶бһ“бҹ”

**бһҹбҹҶбһҒбһ¶бһ“бҹӢ** : бһҳбһ»бһҒбһ„бһ¶бһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҹбһ¶бһҡвҖӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһ“бһҪбһ“бһҹбһүбҹ’бһүбһ¶бһ”бһ¶бһҸбһҳбһ¶бһ“аёӯаёўаё№а№ҲбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбҹ’бһҡбһ¶бһ”бҹӢбҹ”


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**бһ“бһ·бһҷбһҳбһ“бҹҗбһҷ** 

бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ“бҹғбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҳбһ¶бһ“бһҮбһ¶бһ“бҹӢбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһҳбһҪбһҷбҹ—бһҳбһ¶бһ“бһўбһҗбҹҒбһҡбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бҹ” бһ‘бһёбһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһ”бҹҶбһ”бҹӮбһҖбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҡбҹҖбһ”бһҡбһҷ:

**бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһҖбҹҶбһ–бһјбһӣбһ“бҹғбһўбһ“бһ»бһӮбһҳбһ“бҹҚбҹ–**

`name` - бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹғбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһІбҹ’бһҷбһҸбҹ’бһҡбһјбһңбһ бҹ…бҹ” 

`description` - бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһ–бһҺбҹҢбһ“бһ¶бһўбҹҶбһ–бһёбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһңбһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢ 

`parameters` - бһ”бһүбҹ’бһҮбһёбһҸбһҳбҹ’бһӣбҹғбһ“бһ·бһ„бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһІбҹ’бһҷбһҳбҹүбһјбһҠбҹӮбһӣбһ•бһӣбһ·бһҸбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҡбһ”бһҹбҹӢбһңбһ¶ 

**бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһңбһҸбҹ’бһҗбһ» Parametersбҹ–**

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғбһңбһҸбҹ’бһҗбһ» parameters (бһҮбһ¶вҖӢбһ’бһҳбҹ’бһҳбһҸбһ¶ "object")

`properties` - бһ”бһүбҹ’бһҮбһёбһҸбһҳбҹ’бһӣбҹғбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҡбһ”бһҹбҹӢбһңбһ¶ 

**бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһ“бҹғ Parameter бһҸбҹӮбһҳбһҪбһҷбҹ—бҹ–**

`name` - бһҠбһ¶бһҖбҹӢбһ”бһ„бҹ’бһ бһ¶бһүбһҠбҹ„бһҷбһҖбһјбһ“бһҹбҹ„бһўбһҗбҹҒбһҡбһ“бҹ„бҹҮбһҠбҹ„бһҷбһ”бҹ’бһҡбһҷбҹ„бһӮ (бһ§. "role", "product", "level")

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғ parameter бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ“бҹҒбҹҮ (бһ§. "string", "number", "boolean") 

`description` - бһ–бһҺбҹҢбһ“бһ¶бһҡбһ”бһҹбҹӢ parameter бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ“бҹҒбҹҮ 

**бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲбһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбҹ–**

`required` - бһўбһ¶бһҡбҹҒбһҡвҖӢбһ”бһүбҹ’бһҮбһёбһҗбһ¶бһҸбһҫ parameter бһҳбһҪбһҷбһҺбһ¶бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһІбҹ’бһҷбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢ 


### бһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ 
бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҡбһҪбһ… бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶ бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ бҹ…бһ‘бҹ… API Chat Completionбҹ” бһҷбһҫбһ„бһ’бҹ’бһңбһҫбһ“бҹҒбҹҮбһҠбҹ„бһҷбһ”бһ“бҹ’бһҗбҹӮбһҳ `functions` бһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҹбҹҶбһҺбһҫбҹ” бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ `functions=functions` бҹ”

бһҖбҹҸбһҳбһ¶бһ“бһҮбһҳбҹ’бһҡбһҫбһҹбһҳбһҪбһҷбһҠбҹӮбһҡ бһҠбһҫбһҳбҹ’бһ”бһёбһҖбҹҶбһҺбһҸбҹӢ `function_call` бһ‘бҹ…бһҮбһ¶ `auto`бҹ” бһ“бҹҒбҹҮбһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһҷбһҫбһ„бһ“бһ№бһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһұбҹ’бһҷ LLM бһҹбһҳбҹ’бһҡбҹҒбһ…бһҗбһ¶бһҸбһҫбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҺбһ¶бһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ… dб»ұaбһӣбһҫбһҹбһ¶бһҡвҖӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһҳбһ·бһ“бһҳбҹӮбһ“бһ”бҹӮбһ„бһ…бҹӮбһҖбһңбһ¶бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ‘бҹҒбҹ”


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҳбһҫбһӣбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ бһҫбһҷбһҳбһҫбһӣбһҗбһ¶бһңбһ¶бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҸбһҳбҹ’бһҡбҹӮбһҸбһҳбҹ’бһҡбһ„бҹӢбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ–

бһҮбһ¶бһҠбҹҶбһ”бһјбһ„ бһҹбһјбһҳбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ бҹ…бһ‘бҹ…бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI бһ бһҫбһҷбһҡбһҖбҹ’бһҹбһ¶бһҹбһ¶бһҡ бһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡбһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮбһҗбһ¶ `response_message`бҹ”

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһ“бһ№бһ„бһ бҹ… Microsoft Learn API бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ¶бһүбһҷбһҖбһ”бһүбҹ’бһҮбһёбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

бһҷбһҫбһ„бһ“бһ№бһ„бһҷбһҖбһўбһ¶бһӮбһ»бһҷбһҳбҹүбһ„бҹӢбһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бһ·бһ„бһ•бҹ’бһӮбһјбһ•бҹ’бһӮбһ„бһңбһ¶бһ‘бҹ…бһұбҹ’бһҷбһўбһ¶бһӮбһ»бһҷбһҳбҹүбһ„бҹӢбһҠбҹӮбһӣбһҳбһҖбһ–бһё LLM бҹ”

бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһҹбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһ“бһ·бһ„бһҸбһҳбҹ’бһӣбҹғбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбһҠбҹ„бһҷбһҹбһ¶бһҡ `search_courses` бҹ” бһңбһ¶бһ“бһ№бһ„бһ•бҹ’бһҸбһӣбҹӢбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҠбҹӮбһӣ LLM бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһё
бһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҸбһ¶бҹ”
