## бһҖбһ¶бһҡбһ”бһҡбһ·бһ…бҹ’бһҶбҹҒбһ‘ 

бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбһ“бһ№бһ„бһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһўбҹҶбһ–бһё: 
- бһҸбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҮбһ¶бһўбҹ’бһңбһё бһ“бһ·бһ„бһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҡбһ”бһҹбҹӢбһңбһ¶ 
- бһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ OpenAI 
- бһҡбһ”бҹҖбһ”бһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё 

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶ 

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҠбһ№бһ„бһңбһ·бһ’бһёбһ’бҹ’бһңбһҫ бһ“бһ·бһ„бһҷбһӣбҹӢбһ–бһё: 

- бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ 
- бһҖбһ¶бһҡбһҸбҹҶбһЎбһҫбһ„бһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҹбҹҒбһңбһ¶ OpenAI 
- бһҡбһ…бһ“бһ¶бһҡбһ”бҹҖбһ”бһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҹбһ·бһ‘бҹ’бһ’бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ 


## бһҖбһ¶бһҡбһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ

бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһ–бһҪбһҖбһҷбһҫбһ„бһ…бһ„бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбҹ’бһҸбһ¶бһҸбһўбһ¶бһ”бҹӢбһўбһ”бҹӢбһҡбҹҶбһ“бҹғбһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶бһҡбһ”бһҹбҹӢбһҷбһҫбһ„ бһҠбҹӮбһӣбһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ” chatbot бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ”бһ…бҹ’бһ…бҹҒбһҖбһ‘бҹҒбһҹбҹ” бһ–бһҪбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһҸбһӣбҹӢбһўбһ“бһ»бһҹбһ¶бһҹбһ“бҹҚбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһҠбҹӮбһӣбһҹбһҳбһҡбһҳбҹ’бһҷбһ“бһ№бһ„бһҖбһҳбҹ’бһҡбһ·бһҸбһҮбҹҶбһ“бһ¶бһү бһҸбһҪбһ“бһ¶бһ‘бһёбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“ бһ“бһ·бһ„бһ”бһ…бҹ’бһ…бҹҒбһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһҠбҹӮбһӣбһ–бһҪбһҖбһӮбҹҒбһҹ tГӯm.

бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҖбһ¶бһҡбһ„бһ¶бһҡбһ“бҹҒбҹҮ бһ–бһҪбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбҹҲ
 - `OpenAI` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҮбһҮбҹӮбһҖбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ
 - `Microsoft Learn Catalog API` бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһҸбһ¶бһҳбһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ
 - `Function Calling` бһҠбһҫбһҳбҹ’бһ”бһёбһҷбһҖбһҹбҹҶбһҺбһҪбһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ бһ бһҫбһҷбһ•бҹ’бһүбһҫбһңбһ¶бһ‘бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҹбҹҶбһҺбһҫ APIбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ бһҷбһҫбһ„бһҳбһҖбһҳбһҫбһӣбһҳбһјбһӣбһ бҹҒбһҸбһ»бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶бһҷбһҫбһ„бһ…бһ„бҹӢбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹҒбҹҮбһҮбһ¶бһӣбһҫбһҖбһҠбҹҶбһ”бһјбһ„бҹ–


### бһҳбһјбһӣбһ бҹҒбһҸбһ»бһҖбһ¶бһҡбһ бҹ…бһўбһ“бһ»бһӮбһҳбһ“бҹҚ  

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһ бҹӮбһӣбһҮбһ¶бһ”бһ¶бһ“бһҷбһӣбҹӢбһўбҹҶбһ–бһёбһўбҹҶбһҺбһ¶бһ…бһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһӮбһҳбҹ’бһҡбҹ„бһ„бһ—бһ¶бһҹбһ¶бһ’бҹҶбһ“бһ¶бһ“бһ¶ (LLMs)бҹ” бһҹбһ„бҹ’бһғбһ№бһҳбһҗбһ¶бһўбҹ’бһ“бһҖбһҖбҹҸбһўбһ¶бһ…бһҳбһҫбһӣбһғбһҫбһүбһҖбҹҶбһҺбһҸбҹӢбһҒбҹ’бһӣбҹҮбҹ—бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһңбһ¶бһ‘бҹ…бһ•бһ„бһҠбҹӮбһҡбҹ”  

бһҖбһ¶бһҡбһ бҹ…бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһӮбһәбһҮбһ¶бһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹбһҳбһҪбһҷбһ“бҹғбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI бһҠбҹӮбһӣбһҡбһ…бһ“бһ¶бһЎбһҫбһ„бһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһ”бһүбҹ’бһ бһ¶бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–  

бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһҳбһ·бһ“бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбҹ–  
- бһҳбһ»бһ“бһ–бҹҒбһӣбһҳбһ¶бһ“бһҖбһ¶бһҡбһ бҹ…бһўбһ“бһ»бһӮбһҳбһ“бҹҚ бһ…бһҳбҹ’бһӣбһҫбһҷбһ–бһёбһӮбһҳбҹ’бһҡбҹ„бһ„бһ—бһ¶бһҹбһ¶бһ’бҹҶбһ“бҹҒбҹҮбһӮбһәбһҳбһ·бһ“бһҡбҹҖбһ”бһ…бҹҶ бһ“бһ·бһ„бһҳбһ·бһ“бһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбҹ” бһўбҹ’бһ“бһҖбһўwickбһ…бһ·бһ“бһҸбҹ’бһҡбһјбһңбһҸбҹӮбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһҹбҹ’бһҳбһ»бһӮбһҹбҹ’бһҳбһ¶бһүбһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹ„бҹҮбһҹбҹ’бһҡбһ¶бһҷбһҡбҹҖбһ„бһҡбһ¶бһӣбҹӢбһ”бһҳбҹ’бһҡбҹӮбһ”бһҳбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһӣбһ‘бҹ’бһ’бһ•бһӣбҹ”  

бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ…бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбҹ–

```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"
```

```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

### бһ’бһ¶бһҸбһ»вҖӢбһ“бҹғвҖӢбһҖбһ¶бһҡбһ бҹ…вҖӢбһҳбһ»бһҒбһ„бһ¶бһҡ 

#### бһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһ–бһёбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ 

бһҮбҹҶбһ бһ¶бһ“бһҠбҹҶбһ”бһјбһ„бһӮбһәбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһ¶бһҡвҖӢбһўбҹ’бһ“бһҖвҖӢбһ”бҹ’бһҡбһҫбҹ” бһңбһ¶бһўбһ¶бһ…бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ•бҹ’бһҠбһӣбҹӢбһҸбһҳбҹ’бһӣбҹғбһІбҹ’бһҷбһЎбһҫбһ„бһңбһ·бһүбһҠбҹ„бһҷ ОҙП…ОҪОұОјО№ОәО¬вҖӢвҖӢбһҠбҹ„бһҷбһҷбһҖбһҸбһҳбҹ’бһӣбҹғбһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ бһ¬вҖӢбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ•бҹ’бһҠбһӣбҹӢбһҸбһҳбҹ’бһӣбҹғбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһ“бҹҒбҹҮбһҮбһ¶бһӣбһҫбһҖбһҠбҹҶбһ”бһјбһ„бһўбҹ’бһ“бһҖбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ Chat Completions API бһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбҹҶбһҺбһҸбҹӢ `role` бһ“бһ·бһ„ `content` бһ“бҹғбһҹбһ¶бһҡбҹ”

`role` бһўбһ¶бһ…бһҮбһ¶бһўбҹ’бһңбһёбһҳбһҪбһҷбһ–бһё `system` (бһ”бһ„бҹ’бһҖбһҫбһҸбһ…бҹ’бһ”бһ¶бһ”бҹӢ), `assistant` (бһҳбҹүбһјбһҠбҹӮбһӣ) бһ¬ `user` (бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ)бҹ” бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһ“бҹҒбҹҮбһҮбһ¶бһҡбҹҖбһ„бһҡбһ бһјбһҸбһҖбҹ’бһ“бһ»бһ„ `user` бһ“бһ·бһ„бһҹбҹҶбһҺбһҪбһҡбһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһҳбһҪбһҷбҹ”


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡ бһ“бһ·бһ„бһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡвҖӢбһҡбһ”бһҹбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹ„бҹҮбҹ” бһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҸбҹӮвҖӢбһҳбһ»бһҒбһ„бһ¶бһҡбһ”бһҳбҹ’бһҡбһ»бҹҮбһҳбһҪбһҷбһҲбҹ’бһҳбҹ„бҹҮбһҮбһ¶ `search_courses` бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһ…бҹ’бһҡбһҫбһ“бһ”бһ¶бһ“бҹ”

**бһҹбҹҶбһҒбһ¶бһ“бҹӢ** : бһҳбһ»бһҒбһ„бһ¶бһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҹбһ¶бһҡвҖӢбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһ бһҫбһҷбһ“бһ№бһ„бһҸбҹ’бһҡбһјбһңбһ”бһүбҹ’бһ…бһјбһӣвҖӢбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһ“бһҪбһ“вҖӢбһҸбһјбһҒбһ·бһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бҹ”


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**бһ“бһ·бһҷбһҳбһ“бҹҗбһҷ**

бһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһ“бҹғбһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһ¶бһ“бһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ“бҹӢбҹ— бһ бһҫбһҷбһ“бһёбһҳбһҪбһҷбҹ—бһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бҹ” бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһ–бһёбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҠбһ¶бһҖбҹӢбһҮбһ¶бһҡбһ”бһ¶бһҷбһҖбһ¶бһҡбһҺбҹҚбҹ–

**бһӣбһҖбҹ’бһҒбһҺбҹҲбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡбһӣбһҫбһҹбҹ’бһҸбһ„бҹӢбһҠбһ¶бһҡбҹ–**

`name` - бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһўбҹ„бһҷбһ бҹ…бҹ”

`description` - бһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһҠбҹӮбһӣбһҳбһ»бһҒбһ„бһ¶бһҡбһ“бҹҒбҹҮбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбҹ” бһҖбҹ’бһ“бһ»бһ„бһ“бҹҒбҹҮбһңбһ¶бһҹбҹҶбһҒбһ¶бһ“бҹӢбһҺбһ¶бһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҷбҹүбһ¶бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһ“бһ·бһ„бһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢбҹ”

`parameters` - бһ”бһүбҹ’бһҮбһёбһҸбһҳбҹ’бһӣбҹғбһ“бһ·бһ„бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһўбҹ„бһҷбһҳбҹүбһјбһҠбҹӮбһӣбһ•бһӣбһ·бһҸбһЎбһҫбһ„бһҖбҹ’бһ“бһ»бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

**бһӣбһҖбҹ’бһҒбһҺбҹҲбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ“бҹғбһңбһҸбҹ’бһҗбһ» Parameters:**

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғбһңбһҸбҹ’бһҗбһ» parameters (бһҮбһ¶бһҡбһҝбһҷа№Ҷ "object")

`properties` - бһ”бһүбҹ’бһҮбһёбһ“бҹғбһҸбһҳбҹ’бһӣбҹғбһ…бҹ’бһ”бһ¶бһҹбҹӢбһӣбһ¶бһҹбҹӢбһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһҳбҹ’бһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”

**бһӣбһҖбҹ’бһҒбһҺбҹҲбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ“бҹғ Parameters бһҠбһ¶бһ…бҹӢбһ“бҹ…бҹ’бһҡбһ“бҹӮбһҖбҹ–**

`name` - бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҠбҹ„бһҷвҖӢбһҹбҹ’бһңбҹҗбһҷбһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·вҖӢбһҠбҹ„бһҷбһӮбһ“бҹ’бһӣбһ№бҹҮбһӣбһҖбҹ’бһҒбһҺбҹҲ (бһ§. "role", "product", "level")

`type` - бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ“бҹғ parameter бһҠбһ¶бһ…бҹӢбһ“бҹ…бҹ’бһҡбһ“бҹӮбһҖбһ“бҹҒбҹҮ (бһ§. "string", "number", "boolean")

`description` - бһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһ“бҹғ parameter бһҠбһ¶бһ…бҹӢбһ“бҹ…бҹ’бһҡбһ“бҹӮбһҖбһ“бҹҒбҹҮ

**бһӣбһҖбҹ’бһҒбһҺбҹҲбһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбҹ–**

`required` - бһҮбһ¶бһҹбҹҶбһҺбһ»бҹҶбһҡбһ¶бһҷбһ“бһ¶бһҳбһ”бһүбҹ’бһҮбһёбһҗбһ¶бһҸбһҫ parameter бһҳбһҪбһҷбһҺбһ¶бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһұбҹ’бһҷбһ”бһүбҹ’бһ…бһ”бҹӢ

бһҳбһ¶бһ“бһҮбһҳбҹ’бһҡбһҫбһҹбһҳбһҪбһҷбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ бҹ…бһҗбһ¶ `function_call` бһҮбһ¶ `auto`бҹ” бһңбһ¶бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһҷбһҫбһ„бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷ LLM бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҳбһ»бһҒбһ„бһ¶бһҡвҖӢбһҠбҹӮбһӣбһӮбһҪбһҡбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ…бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҹбһ¶бһҡвҖӢбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ бһҳбһ·бһ“бһҳбҹӮбһ“бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҠбҹ„бһҷбһҷбһҫбһ„бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһ“бҹ„бҹҮбһ‘бҹҒбҹ”

## 3. бһҖбһ¶бһҡбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһҖбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷбҹ” 


бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһҷбһҫбһ„бһ”бһ¶бһ“бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһ…бһҳбҹ’бһӣбһҫбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбҹҖбһ”бһ…бҹҶбһ–бһё LLM бһ”бһ¶бһ“бһҡбһҪбһ…бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһўбһ¶бһ…бһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷбһ”бһ¶бһ“бҹ” 

### бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһӣбҹҶбһ бһјбһҡ 

бһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҪбһҳбһ”бһүбҹ’бһ…бһјбһӣбһңбһ¶бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бһҳбһҖбһҷбһҫбһ„бһ’бҹ’бһңбһҫбһҸбһ¶бһҳбһҮбҹҶбһ бһ¶бһ“бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ– 

бһҠбҹҶбһ”бһјбһ„бһҳбһҖбһҷбһҫбһ„бһ бҹ…бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI бһ бһҫбһҷбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһҹбһ¶бһҡвҖӢбһҖбҹ’бһ“бһ»бһ„бһўбһҗбҹҒбһҡвҖӢбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ `response_message`бҹ” 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһҖбҹҶбһҺбһҸбҹӢбһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹӮбһӣбһ“бһ№бһ„бһ бҹ… Microsoft Learn API бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ”бһүбҹ’бһҮбһёбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бҹ– 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

бһҮбһ¶бһ‘бҹҶбһ“бҹҖбһҳбһ‘бҹҶбһӣбһ¶бһ”бҹӢбһӣбҹ’бһўбһ”бҹҶбһ•бһ»бһҸ бһҷбһҫбһ„бһ“бһ№бһ„бһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҳбҹүбһјбһҢбҹӮбһӣбһ…бһ„бҹӢбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһ¬бһ‘бҹҒбҹ” бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһ»бһҒбһ„бһ¶бһҡбһҳбһҪбһҷбһҖбҹ’бһ“бһ»бһ„бһ…бҹҶбһҺбҹ„бһҳбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҳбһ¶бһ“ бһ бһҫбһҷбһ—бҹ’бһҮбһ¶бһ”бҹӢбһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҳбһ»бһҒбһ„бһ¶бһҡбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ бҹ…бҹ” 
бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ“бһ№бһ„бһҷбһҖбһўбһ¶бһӮбһ»бһҷбһҳбҹүбһ„бҹӢбһ“бҹғбһҳбһ»бһҒбһ„бһ¶бһҡ бһ бһҫбһҷбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһўбһ¶бһӮбһ»бһҷбһҳбҹүбһ„бҹӢбһ–бһё LLMбҹ”

бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһҹбһ¶бһҡбһ бҹ…бһҳбһ»бһҒбһ„бһ¶бһҡ бһ“бһ·бһ„бһҸбһҳбҹ’бһӣбҹғбһҠбҹӮбһӣбһ”бһ¶бһ“бһҸбҹ’бһҡбһЎбһ”бҹӢбһ–бһёбһҹбһ¶бһҡ `search_courses` бҹ” бһ“бҹҒбҹҮбһ•бҹ’бһҸбһӣбҹӢбһІбҹ’бһҷ LLM бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҠбҹӮбһӣбһңбһ¶бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһё
аё•аёӯаёҡаёҒаёҘаёұаёҡбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҸбһ¶бҹ” 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһҹбһ¶бһҡвҖӢбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ“бҹ’бһ‘бһ¶бһ“бҹӢбһҹбһҳбҹҗбһҷбһ‘бҹ…бһҖбһ¶бһ“бҹӢ LLM бһҠбһҫбһҳбҹ’бһ”бһёбһҷбһҫбһ„бһўбһ¶бһ…бһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷбһ—бһ¶бһҹбһ¶бһ”бҹӮбһ”бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ· бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһ“бһ№бһ„бһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷ JSON бһҡбһ”бһҹбҹӢ APIбҹ”
