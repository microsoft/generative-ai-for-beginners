## бһӮбҹ„бһӣбһҠбҹ…бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶

бһ“бҹ…бһ…бһ»бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…:

- бһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһӮбҹҶбһҡбһј `gpt-image` бһ“бһ·бһ„бһ—бһ¶бһ–бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһҡбһңбһ¶бһ„бһӮбҹҶбһҡбһјбһ“бҹҒбҹҮ бһ“бһ·бһ„бһӮбҹҶбһҡбһј DALLВ·E бһ…бһ¶бһҹбҹӢбҹ—бҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбҹӮбһҡбһјбһ”бһ—бһ¶бһ–бҹ”

## бһҸбһҫбһўбҹ’бһңбһёбһ‘бҹ…бһҮбһ¶бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–?

бһӮбҹҶбһҡбһјбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҳбһҖбһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бҹ” бһӮбҹҶбһҡбһјбһҹбһҳбҹҗбһҷбһҗбҹ’бһҳбһёбһҠбһјбһ…бһҮбһ¶ `gpt-image` бһҹбһ·бһҖбҹ’бһҹбһ¶бһҒбҹ’бһҹбҹӮбһ‘бҹҶбһ“бһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„бһҡбһңбһ¶бһ„бһўбһҸбҹ’бһҗбһ”бһ‘ бһ“бһ·бһ„бһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһ–бҹҒбһӣбһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһңбһ¶бһ”бҹҶбһ”бҹӮбһҖбһҹбҹҶбһЎбҹҒбһ„бһ…бҹғбһҠбһ“бҹ’бһҷбһ‘бҹ…бһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ•бҹ’бһӮбһјбһ•бҹ’бһҹбһ–бҹ’бһңбһӮбҹ’бһ“бһ¶бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

бһӮбҹҶбһҡбһјбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһӣбҹ’бһ”бһёбһҲбҹ’бһҳбҹ„бҹҮбһ–бһёбһҡбһ”бҹ’бһҡбһ—бҹҒбһ‘бһӮбһә:

- **`gpt-image` (OpenAI)** вҖ” бһҮбһ¶бһ‘бҹҶбһ“бһҫбһ”бһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“ бһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбҹ” бһңбһ¶бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҖбһ¶бһҡбһ”бһҳбҹ’бһӣбҹӮбһ„бһўбһҸбҹ’бһҗбһ”бһ‘бһ‘бҹ…бһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣбһҡбһјбһ”бһ—бһ¶бһ– (inpainting бһҮбһ¶бһҳбһҪбһҷбһҳбҹүбһ¶бһҹбҹӢ)бҹ”
- **Midjourney** вҖ” бһӮбҹҶбһҡбһјбһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҮбһ¶бһ”бҹ’бһҡбһ·бһҷбһ—бһ¶бһ–бһ–бһёбһ—бһ¶бһӮбһёбһ‘бһёбһ”бһё бһҳбһ¶бһ“бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһ•бҹ’бһ‘бһ¶бһӣбҹӢ бһ“бһ·бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҖбҹ’бһ“бһ»бһ„ Discordбҹ”

> бһӮбҹҶбһҡбһјбһҡбһјбһ”бһ—бһ¶бһ– OpenAI бһ…бһ¶бһҹбҹӢбҹ— вҖӢвҖ” **DALLВ·E 2** бһ“бһ·бһ„ **DALLВ·E 3** вҖ” бһӮбһәбһҮбһ¶бһҡбһ”бһҹбҹӢбһ…бһ¶бһҹбҹӢбҹ” DALLВ·E 3 бһҳбһ·бһ“бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҠбһ¶бһҖбҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҗбҹ’бһҳбһёбҹ—бһ‘бҹҒ бһ бһҫбһҷбһҳбһ»бһҒбһ„бһ¶бһҡ `create_variation` бһҳбһ¶бһ“бһҸбҹ’бһҡбһ№бһҳбһҸбҹӮ DALLВ·E 2 бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ” бһҹбһјбһҳбһ”бҹ’бһҡбһҫбһӮбҹҶбһҡбһј `gpt-image` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҗбҹ’бһҳбһёбҹ”

бһ“бҹ… Microsoft Foundry, **`gpt-image-2`** бһӮбһәбһҮбһ¶бһӮбҹҶбһҡбһјбһҡбһјбһ”бһ—бһ¶бһ–бһҗбҹ’бһҳбһё бһ“бһ·бһ„бһҳбһ¶бһ“бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҒбҹ’бһ–бһҹбҹӢбһ”бҹҶбһ•бһ»бһҸ бһ бһҫбһҷбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ•бҹ’бһҸбһӣбҹӢбһўбһ“бһ»бһҹбһ¶бһҹбһ“бҹҚбһұбҹ’бһҷбһ”бҹ’бһҡбһҫбһҮбһ¶бһҠбһҫбһҳбҹ” `gpt-image-1.5` бһ“бһ·бһ„ `gpt-image-1-mini` бһҖбҹҸбһҳбһ¶бһ“бһҹбҹ’бһҡбһ¶бһ”бҹӢбһҠбҹ„бһҷбһ‘бһјбһ‘бҹ…бһ•бһ„бһҠбҹӮбһҡбҹ”

> **бһҹбҹҶбһҒбһ¶бһ“бҹӢбҹ–** бһӮбҹҶбһҡбһј `gpt-image` бһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ‘бҹ…бһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ“бһ·бһҳбҹ’бһҳбһ·бһҸбһҮбһ¶бһҡбһ¶бһ„ **base64** (`b64_json`), бһҳбһ·бһ“бһҳбҹӮбһ“бһҮаёІаёҘбһёбһ„ URL бһ‘бҹҒбҹ” бһҖбһјбһҠбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҖбһјбһҠбһўбһјбһҹбһҒбҹ’бһҹбҹӮбһўбһҸбҹ’бһҗбһ”бһ‘ base64 бһ‘бҹ…бһҮбһ¶ bytes бһ бһҫбһҷбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖ бһңбһ¶бһҳбһ·бһ“бһҳбһ¶бһ“ URL бһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ¶бһүбһҷбһҖбһ‘бҹҒбҹ”


## бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһҸбһҫбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–? бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- **python-dotenv**, бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҺбҹӮбһ“бһ¶бҹҶбһҷбҹүбһ¶бһ„бһҒбҹ’бһӣбһ¶бҹҶбһ„бһІбҹ’бһҷбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮ бһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҖбҹ’бһҹбһ¶бһҜбһҖбһҮбһ“бһ—бһ¶бһ–бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ *.env* бһҶбҹ’бһ„бһ¶бһҷбһ–бһёбһҖбһјбһҠбҹ”
- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһҮбһ¶бһҠбһҫбһҳбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёвҖӢбһ‘бҹҶбһ“бһ¶бһҖбҹӢбһ‘бҹҶбһ“бһ„бһҮбһ¶бһҳбһҪбһҷ OpenAI APIбҹ”
- **pillow**, бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ Pythonбҹ”

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһҡбһҪбһ…бһ‘бҹҒ бһҹбһјбһҳбһ’бҹ’бһңбһҫбһҸбһ¶бһҳбһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶбһ“бҹ…бһӣбһҫбһ‘бҹҶбһ–бҹҗбһҡ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ’бһ“бһ’бһ¶бһ“ Microsoft Foundry бһ“бһ·бһ„бһҳбҹүбһјбһҢбҹӮбһӣбҹ” бһҮбҹ’бһҡбһҫбһҹ **gpt-image-2** бһҮбһ¶бһҳбҹүбһјбһҢбҹӮбһӣ (бһҳбҹүбһјбһҢбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ– Azure OpenAI бһҗбҹ’бһҳбһёбһ”бҹҶбһ•бһ»бһҸ; DALLВ·E бһӮбһәбһҮбһ¶бһҳбҹүбһјбһҢбҹӮбһӣбһ…бһ¶бһҹбҹӢ)бҹ”

    import base64

            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # бһ§. "gpt-image-2"

        image_dir = os.path.join(os.curdir, 'images')

        image_path = os.path.join(image_dir, 'generated-image.png')

        # бһҳбҹүбһјбһҢбҹӮбһӣ gpt-image бһ”бһүбҹ’бһҮбһјбһ“бһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһӮбҹ„бһӣбһҠбҹ… base64 (b64_json) бһҳбһ·бһ“бһҳбҹӮбһ“ URL бһ‘бҹҒ

        image = Image.open(image_path)
        image.show()

    # бһ…бһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһҒбһ»бһҹбһ”бҹ’бһӣбҹӮбһҖ
    except BadRequestError as err:

### бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ”бһ“бҹ’бһҗбҹӮбһҳбһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–

бһҳбһҖбһҳбһҫбһӣбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡбһ“бҹғ `images.generate`бҹ–

- **prompt**, бһӮбһәбһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ“бҹғбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһңбһ¶бһӮбһә "бһҖбҹ’бһҡбһ–бһҫбһӣбһҫбһҹбҹҒбҹҮ бһҖбһ¶бһ“бҹӢбһҲбһ»бһҸбһҖбҹ’бһҡбһЎбһ»бһҖ бһ“бҹ…бһ…бҹҶбһӣбһҫбһҷбһ–бһ–бһҖ бһҠбҹӮбһӣбһҳбһ¶бһ“бһ•бҹ’бһҖбһ¶ daffodils бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“"бҹ”
- **size**, бһӮбһәбһҮбһ¶ бһ‘бҹҶбһ бҹҶбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸ (бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `1024x1024`, `1536x1024`, `1024x1536`, бһ¬ `"auto"` )бҹ”
- **n**, бһӮбһәбһҮбһ¶бһ…бҹҶбһ“бһҪбһ“бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһҪбһҷбһҡбһјбһ”бҹ”
- **model**, бһӮбһәбһҮбһ¶бһҲбҹ’бһҳбҹ„бҹҮбһҖбһ¶бһҡбһҸбҹҶбһЎбһҫбһ„бһҳбҹүбһјбһҢбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ (бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `gpt-image-2`)бҹ”

> бһҳбҹүбһјбһҢбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһҳбһ·бһ“бһҳбһ¶бһ“бһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡ `temperature` бһ‘бҹҒ вҖ” бһңбһ¶бһҮбһ¶бһҖбһ¶бһҡбһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһ—бһ¶бһ–бһ…бһҳбҹ’бһҡбһ»бҹҮ, бһ бҹ… API бһҳбҹ’бһҸбһ„бһ‘бҹҖбһҸ; бһҠбһҫбһҳбҹ’бһ”бһёбһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһ—бһ¶бһ–бһ…бһҳбҹ’бһҡбһ»бҹҮ, бһ’бҹ’бһңбһҫбһІбҹ’бһҷ prompt бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ…бҹ’бһ”бһ¶бһҹбҹӢбҹ”

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһҳбһҪбһҷбһҒбҹ’бһҹбҹӮ Python бһ…бҹҶбһ“бһҪбһ“бһ”бҹүбһ»бһ“бҹ’бһҳбһ¶бһ“бҹ” бһҳбҹүбһјбһҢбҹӮбһӣ `gpt-image` бһҖбҹҸбһўбһ¶бһ… **бһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣ** бһҡбһјбһ”бһ—бһ¶бһ–бһҳбһ¶бһ“ бһҡбһҪбһ…бһ бһҫбһҷбһ”бһ¶бһ“бһ•бһ„бһҠбҹӮбһҡбҹ” бһҠбҹ„бһҷбһ•бҹ’бһҠбһӣбҹӢбһҡбһјбһ”бһ—бһ¶бһ– бһҳбһҪбһҷ **mask** бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹ (бһҠбҹӮбһӣбһҹбһҳбҹ’бһӮбһ¶бһӣбҹӢбһҸбҹҶбһ”бһ“бҹӢбһҠбҹӮбһӣбһ…бһ„бҹӢбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҠбһјбһҡ), бһ“бһ·бһ„ prompt бһҳбһҪбһҷ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҠбһјбһҡбһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғбһҡбһјбһ”бһ—бһ¶бһ– вҖ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһҪбһҖбһҠбһ¶бһҖбҹӢбһӣбһҫбһҖбҹ’бһҡбһ–бһҫбһҡбһ”бһҹбҹӢбһҷбһҫбһ„бҹ”
