- DALL-E бһ“бһ·бһ„ Midjourney, бһҸбһҫбһңбһ¶бһўбҹ’бһңбһё бһ“бһ·бһ„бһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”
- бһҸбһҫбһўбҹ’бһ“бһҖбһӮбһҪбһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҹбһ·бһҖбҹ’бһҹбһ¶

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–

- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһ–бҹ’бһҡбҹҶбһҠбҹӮбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһҳбһ»бһҒбһ”бһ»бһ–бҹ’бһңбһ”бһ‘бҹ”
- бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ DALL-E бһ“бһ·бһ„ Midjourneyбҹ”

## бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶вҖӢбһӮбһҪбһҡбһ”бһ„бҹ’бһҖбһҫбһҸвҖӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёвҖӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–?

бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһӮбһәбһҮбһ¶бһңбһ·бһ’бһёбһҠбҹҸбһӣбҹ’бһўбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҡбһ”бһҹбҹӢ AI бһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹ’бһңбһёбһҗбҹ’бһҳбһёбҹ” бһңбһ¶бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбҹ–

- **бһҖбҹӮбһ”бҹ’бһҡбҹӮбһӣбҹ’бһ”бһҫбһҡбһ“бһ·бһ„бһҹбһҳбһ¶бһҹбһ—бһ¶бһ–бһҡбһјбһ”бһ—бһ¶бһ–**бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ•бҹ’бһҹбҹҒбһ„бҹ— бһҠбһјбһ…бһҮбһ¶ бһҖбһ¶бһҡбһҖбҹӮбһҡбһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбһ¶бһҡбһҹбһҳбһ¶бһҹбһ—бһ¶бһ–бһҡбһјбһ”бһ—бһ¶бһ–бҹ”

- **бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһ§бһҹбҹ’бһҹбһ¶бһ бһҖбһҳбҹ’бһҳбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“**бҹ” бһңбһ¶бһҖбҹҸбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ”бһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһ§бһҹбҹ’бһҹбһ¶бһ бһҖбһҳбҹ’бһҳбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ Medtech, бһ‘бҹҒбһҹбһ…бһҡбһҺбҹҚ, бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”

## бһҹбҹ’бһҗбһ¶бһ“бһ—бһ¶бһ–бҹ– Edu4All

бһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғбһҳбҹҒбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ“бҹ’бһҸбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ Edu4Allбҹ” бһҹбһ·бһҹбҹ’бһҹбһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒбҹ” бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһЎбһҫбһ„бһ–бһ·бһҸбһҮбһ¶бһҹбҹ’бһҗбһ·бһҸбһ“бҹ…бһӣбһҫбһҮбһҳбҹ’бһҡбһҫбһҹбһҡбһ”бһҹбҹӢбһҹбһ·бһҹбҹ’бһҹ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһңбһ¶бһўбһ¶бһ…бһҮбһ¶бһҡбһјбһ”бһҗбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бҹ’бһҡбҹ„бһҹбһӣбһ¶бһ—бһ¶бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒ бһ¬бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбһҪбһўбһ„бҹ’бһӮбһҗбҹ’бһҳбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒ бһ¬бһҮбһҪбһҷбһ–бһҪбһҖбһӮбҹҒбһҳбһҫбһӣбһғбһҫбһүбһӮбҹҶбһ“бһ·бһҸбһ“бһ·бһ„бһңбһёбһҹбҹҗбһҷбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒбҹ”

## DALL-E бһ“бһ·бһ„ Midjourney бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһҠбҹҶбһ”бһјбһ„ [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) бһӮбһәбһҮбһ¶бһҳбҹүбһјбһҠбҹӮбһӣ Generative AI бһҠбҹӮбһӣбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳ transformer бһҮбһ¶бһҳбһҪбһҷ *autoregressive transformer* бҹ”

*Autoregressive transformer* бһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶вҖӢбһўбһҸбҹ’бһҗбһ”бһ‘ бһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ·бһҹбҹ’бһҹбһӣбһҫбһҖбһҳбһҪбһҷ бһ бһҫбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ”бҹ’бһҡбһҫбһ”бһ·бһҹбҹ’бһҹбһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһ бһҫбһҷ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ·бһҹбҹ’бһҹбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ” бһңбһ¶бһҶбҹ’бһӣбһ„бһҖбһ¶бһҸбҹӢбһҮбһ¶бһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҖбҹ’бһ“бһ»бһ„бһ”бһҺбҹ’бһҸбһ¶бһүбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бҹ’бһҡбһҹбһ¶бһ‘ бһҡбһ бһјбһҸбһҠбһӣбҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбҹ”

бһҮбһ¶бһҳбһҪбһҷбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ“бҹҒбҹҮ DALL-E бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ”бҹӮбһ”бһ”бһ‘ бһңбһҸбҹ’бһҗбһ» бһӣбһҖбҹ’бһҒбһҺбҹҲ бһ“бһ·бһ„бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбһҖбҹ’бһ“бһ»бһ„бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҖбҹ’бһҡбҹ„бһҳ DALL-E 2 бһ“бһ·бһ„ 3 бһўбһ¶бһ…бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһҳбһ»бһ“бҹ”

            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһҹбҹҶбһҺбһҪбһҡбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ
            size='1024x1024',

        image_dir = os.path.join(os.curdir, 'images')
    
        # бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҗбһҸбһҳбһ·бһ“бһҳбһ¶бһ“ бһҹбһјбһҳбһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ¶
        if not os.path.isdir(image_dir):

        generated_image = requests.get(image_url).content  # бһ‘бһ¶бһүбһҷбһҖбһҡбһјбһ”бһ—бһ¶бһ–
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # бһ”бһ„бҹ’бһ бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҫбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ
        image = Image.open(image_path)
        image.show()
    
    # бһ…бһ¶бһ”бҹӢбһ”бһүбҹ’бһ бһ¶бһҖбһҡбһҺбһёбһ”бһүбҹ’бһ бһ¶
    except openai.InvalidRequestError as err:
