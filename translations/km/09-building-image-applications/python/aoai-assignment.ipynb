# бһҖбһ¶бһҡбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–  

бһҳбһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ LLMsбҹ” бһңбһ¶бһҖбҹҸбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһ¶бһ“бһ•бһ„бһҠбҹӮбһҡбҹ” бһҖбһ¶бһҡбһҳбһ¶бһ“бһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҳбһҪбһҷбһўбһ¶бһ…бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһҒбҹ’бһӣбһ¶бҹҶбһ„бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһҹбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ…бһ¶бһ”бҹӢбһ–бһё MedTech бһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳ бһҖбһҳбҹ’бһҹбһ¶бһ“бҹ’бһҸ бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ бһ“бһ·бһ„бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбҹ” бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһҹбһ·бһҖбҹ’бһҹбһ¶бһўбҹҶбһ–бһёбһҳбҹүбһјбһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бҹҒбһүбһ“бһ·бһҷбһҳбһ–бһёбһҡбһҠбҹӮбһӣбһӮбһә DALL-E бһ“бһ·бһ„ Midjourneyбҹ”  

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ  

бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҠбһ”бҹӢбҹ–  

- бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶бһңбһ¶бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбҹ”  
- DALL-E бһ“бһ·бһ„ Midjourney, бһҸбһҫбһңбһ¶бһўбҹ’бһңбһё, бһ бһҫбһҷбһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”  
- бһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ”  

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҹбһ·бһҖбҹ’бһҹбһ¶  

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–  

- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ”  
- бһҖбҹҶбһҺбһҸбҹӢбһ–бҹ’бһҡбҹҶбһҠбҹӮбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҮбһ¶бһҳбһҪбһҷ meta promptsбҹ”  
- бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ DALL-E бһ“бһ·бһ„ Midjourneyбҹ”  

## бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–?  

бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһӮбһәбһҮбһ¶ бһңбһ·бһ’бһёбһ–бһ·бһҹбҹҒбһҹбһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһ–бһёбһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһҡбһ”бһҹбҹӢ Generative AIбҹ” бһңбһ¶бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ, бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–  

- **бһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣ бһ“бһ·бһ„бһҹбҹ’бһҸбһ·бһ…бһҹбҹҠбһёбһҡбһјбһ”бһ—бһ¶бһ–**бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһ•бҹ’бһҹбҹҒбһ„бҹ— бһҠбһјбһ…бһҮбһ¶бһҖбһ¶бһҡбһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣ бһ“бһ·бһ„бһҹбҹ’бһҸбһ·бһ…бһҹбҹҠбһёбһҡбһјбһ”бһ—бһ¶бһ–бҹ”  

- **бһўбһ“бһ»бһңбһҸбҹ’бһҸбһ‘бҹ…бһңбһ·бһҹбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“**бҹ” бһңбһ¶бһҖбҹҸбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһңбһ·бһҹбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ Medtech бһҖбһҳбҹ’бһҹбһ¶бһ“бҹ’бһҸ бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ бһ“бһ·бһ„бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбҹ”

## DALL-E бһ“бһ·бһ„ Midjourney бһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҸбһ·бһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…  

бһҮбһ¶бһҳбһ»бһ“бһҹбһ·бһ“, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) бһҮбһ¶бһҳбҹүбһјбһҠбҹӮбһӣ Generative AI бһҠбҹӮбһӣбһҳбһ¶бһ“бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһӣбһҫбһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳ Transformer бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„ *autoregressive transformer*бҹ”  

*Autoregressive transformer* бһҖбҹҶбһҺбһҸбҹӢбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһҳбҹүбһјбһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһўбһҸбҹ’бһҗбһ”бһ‘ бһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбһ–бһ·бһ…бһҹбҹӮбһӣбһҳбһҪбһҷбһ–бҹҒбһӣ бһ бһҫбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹ„бҹҮбһ”бҹ’бһҡбһҫбһ–бһ·бһ…бһҹбҹӮбһӣбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ–бһ·бһ…бһҹбҹӮбһӣбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ” бһңбһ¶бһҶбҹ’бһӣбһ„бһҖбһ¶бһҸбҹӢбһҮбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҮбһ¶бһ“бҹӢбһҖбҹ’бһ“бһ»бһ„бһ”бһҺбҹ’бһҸбһ¶бһүбһ”бҹ’бһҡбһҹбһ¶бһ‘бһҡбһ бһјбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бҹҒбһүбһӣбҹҒбһүбҹ”  

бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ“бҹҒбҹҮ DALL-E бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһӮбһ»бһҺбһӣбһҖбҹ’бһҒбһҺбҹҲ бһӮбҹ’бһҡбһҝбһ„бһ•бҹ’бһҹбҹҶ бһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹ бһ“бһ·бһ„бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ‘бҹ„бҹҮбһҮбһ¶бһҷбҹүбһ¶бһ„бһҺбһ¶ DALL-E 2 бһ“бһ·бһ„ 3 бһҳбһ¶бһ“бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһӣбһҫбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһ“бҹҒбҹҮбҹ”  


## бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһҸбһҫбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–? бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- **python-dotenv**, бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҺбҹӮбһ“бһ¶бҹҶбһІбҹ’бһҷбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҖбҹ’бһҹбһ¶вҖӢбһҖбһ¶бһҡбһҹбһҳбҹ’бһ„бһ¶бһҸбҹӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ *.env* бһҠбҹ„бһҷбһҶбҹ’бһ„бһ¶бһҷбһ–бһёбһҖбһјбһҠбҹ”
- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷ API OpenAIбҹ”
- **pillow**, бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ Pythonбҹ”
- **requests**, бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫ HTTPбҹ”

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҳбһ·бһ“бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһҡбһҪбһ…бһ‘бҹҒ бһҹбһјбһҳбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶбһ“бҹ…бһӣбһҫбһ‘бҹҶбһ–бҹҗбһҡ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһ’бһ“бһ’бһ¶бһ“ Azure OpenAI бһ“бһ·бһ„бһҳбҹүбһјбһҠбҹӮбһӣбҹ” бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹ **gpt-image-1** бһҮбһ¶ бһҳбҹүбһјбһҠбҹӮбһӣ (бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ– Azure OpenAI бһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбһҖбҹ’бһҸбһ¶бһ“бһ»бһ–бһӣбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“; DALL-E 3 бһӮбһәбһҮбһ¶бһҳбҹүбһјбһҠбҹӮбһӣбһ…бһ¶бһҹбҹӢбһ бһҫбһҷбһҳбһ·бһ“бһҳбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҠбһ¶бһҖбҹӢбһұбҹ’бһҷбһ”бҹ’бһҡбһҫбһҗбҹ’бһҳбһёбһ‘бҹҒ)бҹ”

1. бһ”бһ„бҹ’бһҖбһҫбһҸбһҜбһҖбһҹбһ¶бһҡ *.env* бһҠбҹ„бһҷбһҳбһ¶бһ“бһҳбһ¶бһҸбһ·бһҖбһ¶бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

    бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһ–бҹҗбһҸбҹҚбһҳбһ¶бһ“бһ“бҹҒбҹҮбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбҹҒбһ”бһҹбһ¶бһҷ Microsoft Foundry бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бһ“бһ’бһ¶бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһ•бҹ’бһ“бҹӮбһҖ "Deployments"бҹ”

1. бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһјбһҠбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡвҖӢбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮ *app.py*бҹ–

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # бһ“бһ¶бҹҶбһ…бһјбһӣ dotenv
    dotenv.load_dotenv()

    # бһҖбҹҶбһҺбһҸбҹӢбһўбһҸбһ·бһ—бһјбһҸбһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһ”бһҳбҹ’бһҡбһҫ Azure OpenAI service client
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ API бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһ„бҹ’бһӮбһ¶бһ”бҹӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # бһҖбҹҶбһҺбһҸбҹӢбһҗбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖ
        image_dir = os.path.join(os.curdir, 'images')
    
        # бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҗбһҸбһҳбһ·бһ“бһҳбһ¶бһ“ бһҹбһјбһҳбһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ¶
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһ•бҹ’бһӣбһјбһңбһҡбһјбһ”бһ—бһ¶бһ– (бһӮбһҪбһҡбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһҗбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҜбһҖбһҹбһ¶бһҡбһҸбҹ’бһҡбһјбһңбһҮбһ¶ png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # бһ‘бһ¶бһүбһҷбһҖбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸ
        image_url = generation_response.data[0].url  # бһ‘бһ¶бһүбһҷбһҖ URL бһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”
        generated_image = requests.get(image_url).content  # бһ…бһ»бҹҮбһ‘бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # бһ”бһ„бҹ’бһ бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҫбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ
        image = Image.open(image_path)
        image.show()
    
    # бһ…бһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһҖбҹҶбһ бһ»бһҹ
    except openai.BadRequestError as err:
        print(err)

    ```

### бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ”бһ“бҹ’бһҗбҹӮбһҳбһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–

- **prompt**, бһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһүбҹ’бһ…бһјбһӣбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ” бһҖбҹ’бһ“бһ»бһ„бһҖбһҡбһҺбһёбһ“бҹҒбҹҮ бһҷбһҫбһ„бһҖбҹҶбһ–бһ»бһ„бһ”бҹ’бһҡбһҫ prompt "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils"бҹ”
