# бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– (OpenAI)

бһҳбһ¶бһ“бһҡбһҝбһ„бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ LLMsбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ•бһ„бһҠбҹӮбһҡбҹ” бһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһҳбһјбһҠбһ¶бһӣбҹӢбһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһҹбҹҗбһҷбһңбҹҒбһҮбҹ’бһҮбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһ”бһ…бҹ’бһ…бҹҒбһҖбһңбһ·бһ‘бҹ’бһҷбһ¶ (MedTech), бһҹбҹҶбһ“бһ„бҹӢбһҹбҹҶбһҺбһ№бһҖ, бһ‘бҹҒбһҹбһ…бһҡбһҺбҹҚ, бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ, бһ•бҹ’бһҹбһ–бҹ’бһңбһ•бҹ’бһҹбһ¶бһҷ бһ“бһ·бһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбҹ” бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһҷбһҫбһ„ бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҮбһ¶бһҳбһҪбһҷбһҳбҹүбһјбһҠбҹӮбһӣ **GPT Image** бһҹбһҳбҹҗбһҷбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһ“бҹ…бһӣбһҫбһңбҹҒбһ‘бһ·бһҖбһ¶ OpenAIбҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҹбһ·бһҖбҹ’бһҹбһ¶

бһ“бҹ…бһ…бһ»бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бҹ–

- бһўСҠСҸСҒбҹҮбһ–бһ·бһ…бһ¶бһҡбһҺбһ¶бһҗбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ”бҹ’бһҡбһ¶бһ”бҹӢбһҳбһ¶бһ“бһўбҹ’бһңбһё бһ“бһ·бһ„бһҖбһ“бҹ’бһӣбҹӮбһ„бһҠбҹӮбһӣбһңбһ¶бһҠбһ¶бһҖбҹӢбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбҹ”
- бһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбһӮбҹ’бһҡбһҪбһҹбһ¶бһҡ `gpt-image` бһ“бһ·бһ„бһ—бһ¶бһ–бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһҡбһңбһ¶бһ„бһңбһ¶бһ“бһ·бһ„бһҳбҹүбһјбһҠбҹӮбһӣ DALLВ·E бһҮбҹҶбһ“бһ¶бһ“бҹӢбһ…бһ¶бһҹбҹӢбҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбҹӮбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ¶бһ“бҹ”

## бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһӮбһәбһҮбһ¶бһўбҹ’бһңбһё?

бһҳбҹүбһјбһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ–бһёбһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ (prompt)бҹ” бһҳбҹүбһјбһҠбҹӮбһӣбһ‘бҹҶбһ“бһҫбһ”бһҠбһјбһ…бһҮбһ¶ `gpt-image` бһҡбҹҖбһ“бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҡбһңбһ¶бһ„бһўбһҸбҹ’бһҗбһ”бһ‘бһ“бһ·бһ„бһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҠбҹҶбһҺбһ¶бһҖбҹӢбһҖбһ¶бһӣбһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһ”бһҺбҹ’бһҸбһ¶бһӣ бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһӣбһ¶бһ”бҹӢбһӣбһҫбһҹбҹҶбһЎбҹҒбһ„бһҡбҹҶбһҒбһ¶бһ“бһ…бҹғбһҠбһ“бҹ’бһҷбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһӮбһәбһҸбҹ’бһҡбһјбһңбһ“бһ№бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

бһӮбҹ’бһҡбһҪбһҹбһ¶бһҡбһҳбҹүбһјбһҠбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҡ бһ–бҹ„бһӣбһӮбһә:

- **`gpt-image` (OpenAI)** вҖ” бһҮбһ¶бһҖбҹҶбһҺбһҫбһҸбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбҹ” бһңбһ¶бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ“бһ·бһ„бһҖбһ¶бһҡбһҖбҹӮбһҸбһҳбҹ’бһҡбһјбһңбһҡбһјбһ”бһ—бһ¶бһ– (бһҮбһҪбһҹбһҮбһ»бһӣбһҮбһ¶бһҳбһҪбһҷбһҳбҹүбһ¶бһҹ)
- **Midjourney** вҖ” бһҳбҹүбһјбһҠбҹӮбһӣбһ”бһёбһ—бһ¶бһӮбһёбһӣбҹ’бһ”бһёбһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһ“бһ·бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ Discordбҹ”

> бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ– OpenAI бһ…бһ¶бһҹбҹӢбҹ— вҖ” **DALLВ·E 2** бһ“бһ·бһ„ **DALLВ·E 3** вҖ” бһӮбһәбһҮбһ¶бһҮбҹҶбһ“бһ¶бһ“бҹӢбһ…бһ¶бһҹбҹӢбҹ” DALLВ·E 3 бһҳбһ·бһ“бһ‘бһ¶бһ“бҹӢбһҳбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһҠбһ¶бһҖбҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҗбҹ’бһҳбһёбһ бһҫбһҷ бһ бһҫбһҷбһӣбһҖбҹ’бһҒбһҺбҹҲ `create_variation` бһҳбһ¶бһ“бһҸбҹӮбһ“бҹ… DALLВ·E 2 бһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ” бһҹбһјбһҳбһ”бҹ’бһҡбһҫбһҳбҹүбһјбһҠбҹӮбһӣ `gpt-image` бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҗбҹ’бһҳбһёбҹ—бҹ”

> **бһҹбҹҶбһҒбһ¶бһ“бҹӢ:** бһҳбҹүбһјбһҠбҹӮбһӣ `gpt-image` бһ”бһүбҹ’бһ…бҹҒбһүбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбһ¶бһҡбһ¶бһ„ **base64** (`b64_json`) бһҳбһ·бһ“бһҳбҹӮбһ“ URL бһ‘бҹҒбҹ” бһҖбһјбһҠбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бһҳбҹ’бһӣбҹӮбһ„бһҒбҹ’бһҹбҹӮбһўбһ¶бһҡбҹүбҹҒ base64 бһ‘бҹ…бһҮбһ¶бһ”бҹғ бһ“бһ·бһ„бһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһңбһ¶ вҖ” бһҳбһ·бһ“бһҳбһ¶бһ“ URL бһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ¶бһүбһҷбһҖбһ‘бҹҒбҹ”


## бһҖбһ¶бһҡбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһҸбһҫбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–? бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ§бһ”бһҖбһҡбһҺбҹҚбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- **python-dotenv**, бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҺбҹӮбһ“бһ¶бҹҶбһҷбҹүбһ¶бһ„бһҒбҹ’бһӣбһ¶бҹҶбһ„бһұбҹ’бһҷбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҖбҹ’бһҹбһ¶бһҡбһҹбһҳбҹ’бһ„бһ¶бһҸбҹӢбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ *.env* бһ…бҹҶбһҖбһҺбҹ’бһҸбһ¶бһӣбһ–бһёбһҖбһјбһҠбҹ”
- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһ”бҹ’бһҡбһҫбһҠбһҫбһҳбҹ’бһ”бһёбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷ OpenAI APIбҹ”
- **pillow**, бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ—бһ¶бһҹбһ¶ Pythonбҹ”


1. бһ”бһ„бҹ’бһҖбһҫбһҸбһҜбһҖбһҹбһ¶бһҡ *.env* бһҮбһ¶бһҳбһҪбһҷбһҳбһ¶бһҸбһ·бһҖбһ¶бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. бһ”бҹ’бһҡбһҳбһјбһӣбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҒбһ¶бһ„бһӣбһҫбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡбһҳбһҪбһҷбһҠбҹӮбһӣбһҳбһ¶бһ“бһҲбҹ’бһҳбҹ„бҹҮ *requirements.txt* бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

    ```text
    python-dotenv
    openai
    pillow
    ```


1. бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһҡбһ·бһҹбҹ’бһҗбһ¶бһ“бһңбҹҒбһҡбһ»бһҶбһјбһўбһ¶бһӣбҹӢбһ бһҫбһҷбһҸбҹҶбһЎбһҫбһ„бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбҹ–


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

1. бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһјбһҠбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ *app.py*:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # бһ“бһ¶бҹҶбһ…бһјбһӣ dotenv
    dotenv.load_dotenv()

    # бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһҸбҹ’бһҗбһ» OpenAI (бһўбһ¶бһ“ OPENAI_API_KEY бһ–бһё .env бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ)
    client = openai.OpenAI()


    try:
        # бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ API бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһүбҹ’бһ…бһјбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ
            size='1024x1024',
            n=1
        )
        # бһҖбҹҶбһҺбһҸбҹӢбһҗбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖбһҡбһјбһ”бһ—бһ¶бһ–
        image_dir = os.path.join(os.curdir, 'images')

        # бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҗбһҸбһҳбһ·бһ“бһҳбһ¶бһ“ бһҹбһјбһҳбһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ¶
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҠбһҫбһҳбһ•бҹ’бһӣбһјбһңбһҡбһјбһ”бһ—бһ¶бһ– (бһҹбһјбһҳбһ…бҹҶбһҺбһ¶бҹҶбһҗбһ¶бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҜбһҖбһҹбһ¶бһҡбһҸбҹ’бһҡбһјбһңбһҸбҹӮ png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # бһҳбҹүбһјбһҢбҹӮбһӣ gpt-image бһҒбҹ’бһ…бһ”бҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖ base64 (b64_json) бһҳбһ·бһ“бһҳбҹӮбһ“бһҮбһ¶ URL
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # бһ”бһ„бҹ’бһ бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҫбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ
        image = Image.open(image_path)
        image.show()

    # бһ…бһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһҖбҹҶбһ бһ»бһҹ
    except openai.BadRequestError as err:
        print(err)

    ```

бһҸбҹ„бҹҮбһ–бһ“бҹ’бһҷбһӣбҹӢбһ–бһёбһҖбһјбһҠбһ“бҹҒбҹҮбҹ–

- бһҮбһ¶бһӣбһҫбһҖбһҠбҹҶбһ”бһјбһ„ бһҷбһҫбһ„бһ“бһ¶бҹҶбһ…бһјбһӣбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһҮбһ¶ OpenAI library, dotenv library, бһҳбҹүбһјбһҢбһ»бһӣ base64, бһ“бһ·бһ„ Pillow libraryбҹ”

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸ client бһҠбҹӮбһӣбһўбһ¶бһ“ API key бһ–бһёбһҜбһҖбһҹбһ¶бһҡ ``.env`` бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

    ```python
    # бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһҸбҹ’бһҗбһ» OpenAI
    client = openai.OpenAI()
    ```

- бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖ бһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹҒбҹҮбҹ–

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһүбҹ’бһ…бһјбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ
        size='1024x1024',
        n=1
    )
    ```

    бһҳбҹүбһјбһҢбҹӮбһӣ `gpt-image` бһ•бҹ’бһҠбһӣбҹӢбһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶ String **base64** бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ `data[0].b64_json`бҹ” бһҷбһҫбһ„ decode бһ‘бҹ…бһҮбһ¶ bytes бһ бһҫбһҷбһҹбһҡбһҹбҹҒбһҡбһ…бһјбһӣбһҜбһҖбһҹбһ¶бһҡ вҖ” бһӮбҹ’бһҳбһ¶бһ“ URL бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ‘бһ¶бһүбһҷбһҖбҹ”

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷ бһҷбһҫбһ„бһ”бһҫбһҖбһҡбһјбһ”бһ—бһ¶бһ– бһ бһҫбһҷбһ”бҹ’бһҡбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҫбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһҹбҹ’бһҸбһ„бҹӢбһҠбһ¶бһҡбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһ бһ¶бһүбһңбһ¶бҹ–

    ```python
    image = Image.open(image_path)
    ```

### бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӣбһҳбҹ’бһўбһ·бһҸбһ”бһ“бҹ’бһҗбҹӮбһҳбһўбҹҶбһ–бһёбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–

бһҸбҹ„бҹҮбһҳбһҫбһӣбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡ бһ“бҹғ `images.generate`бҹ–

- **model** бһҮбһ¶ бһҳбҹүбһјбһҢбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһҫ (бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `gpt-image-1`)бҹ”
- **prompt** бһҮбһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ–бһҺбҹҢбһ“бһ¶бһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһӮбһәбһҮбһ¶ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils"бҹ”
- **size** бһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸ (бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `1024x1024`, `1536x1024`, `1024x1536`, бһ¬ `"auto"`)бҹ”
- **n** бһҮбһ¶бһ…бҹҶбһ“бһҪбһ“бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҳбһҪбһҷбҹ”

> бһҳбҹүбһјбһҢбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһҳбһ·бһ“бһҒбҹ’бһңбһҖбҹӢбһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡ `temperature` бһ‘бҹҒ вҖ” бһңбһ¶бһҮбһ¶бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ¶бһ“бһҖбһ¶бһҡбһ”бҹӮбһ”бһ•бҹ’бһҹбҹҒбһ„бҹ— бһҮбһ¶бһ“бҹӢ API бһҳбҹ’бһҸбһ„бһ‘бҹҖбһҸ; бһҠбһҫбһҳбҹ’бһ”бһёбһҖбһ¶бһҸбҹӢбһ”бһ“бҹ’бһҗбһҷбһҖбһ¶бһҡбһ”бҹӮбһ”бһ•бҹ’бһҹбҹҒбһ„бҹ— бһ”бһ„бҹ’бһҖбһҫбһҸ prompt бһІбҹ’бһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбҹ”

## бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ”бһ“бҹ’бһҗбҹӮбһҳбһ“бҹғбһҖбһ¶бһҡбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһҳбһҪбһҷбһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢ Python бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бҹ” бһҳбҹүбһјбһҢбҹӮбһӣ `gpt-image` бһҖбҹҸбһўбһ¶бһ… **бһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣ** бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһҳбһ¶бһ“бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбҹ” бһҠбҹ„бһҷбһ•бҹ’бһҸбһӣбҹӢбһҡбһјбһ”бһ—бһ¶бһ–, **mask** бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹ (бһҠбҹӮбһӣбһҹбҹҶбһӮбһ¶бһӣбҹӢбһҸбҹҶбһ”бһ“бҹӢбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбҹӮбһ”бҹ’бһҡбҹӮ), бһ“бһ·бһ„ prompt бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһҳбҹ’бһӣбҹӮбһ„бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһҡбһ”бһҹбҹӢбһҡбһјбһ”бһ—бһ¶бһ– - бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһ”бһ“бҹ’бһҗбҹӮбһҳбһҳбһҪбһҖбһ‘бҹ…бһҖбҹҸбһ”бһ¶бһ“бҹ”

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# бһҖбһ¶бһҡбһҖбҹӮбһҹбһҳбҹ’бһҡбһҪбһӣбһҖбҹҸбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ•бҹ’бһүбһҫбһҸбҹ’бһҡбһЎбһ”бҹӢбһҮбһ¶бһҡбһјбһ”бһҳбһ“бҹ’бһҸbase64 бһ•бһ„бһҠбҹӮбһҡ
edited_image = base64.b64decode(response.data[0].b64_json)
```

бһҡбһјбһ”бһ—бһ¶бһ–бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҳбһ¶бһ“бһҸбҹӮбһҳбһҪбһҷбһ‘бһ“бҹ’бһҹбһ¶бһҷ; бһҡбһјбһ”бһ—бһ¶бһ–бһ…бһ»бһ„бһҖбҹ’бһҡбҹ„бһҷбһҳбһ¶бһ“бһҳбһҪбһҖбһӣбһҫбһ‘бһ“бҹ’бһҹбһ¶бһҷбҹ”
