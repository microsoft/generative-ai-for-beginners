# бһҖбһҹбһ¶бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–  

бһҳбһ¶бһ“бһўбҹ’бһңбһёбҹ—бһҮбһ¶бһ„бһҹбһҡбһҹбҹҒбһҡбһўбһҸбҹ’бһҗбһ”бһ‘бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ LLMsбҹ” бһңбһ¶бһҖбҹҸбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһўбһҸбҹ’бһҗбһ”бһ‘бһ•бһ„бһҠбҹӮбһҡбҹ” бһҖбһ¶бһҡбһҳбһ¶бһ“бһҡбһјбһ”бһ—бһ¶бһ–бһҮбһ¶бһҳбһјбһҠбһ¶бһӣбһёбһҸбҹҒбһўбһ¶бһ…бһҳбһ¶бһ“бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһҒбҹ’бһӣбһ¶бҹҶбһ„бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһҹбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ…бһ¶бһ”бҹӢбһ–бһёбһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҮбһёбһңбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡ бһңбһ·бһ…бһ·бһҸбҹ’бһҡбһҹбҹ’бһҗбһ¶бһ”бһҸбҹ’бһҷбһҖбһҳбҹ’бһҳ бһ‘бҹҒбһҹбһ…бһҡбһҺбҹҚ бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ бһ“бһ·бһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбҹ” бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһҹбһ·бһҖбҹ’бһҹбһ¶бһўбҹҶбһ–бһёбһӮбҹҶбһҡбһјбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бҹҒбһүбһ“бһ·бһҷбһҳбһ–бһёбһҡбһҠбһјбһ…бһҮбһ¶ DALL-E бһ“бһ·бһ„ Midjourneyбҹ”  

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ  

бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһӮбҹ’бһҡбһ”бһҠбһҺбҹ’бһҸбһ”бҹӢбҹ–

- DALL-E бһ“бһ·бһ„ Midjourney бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҒбҹ’бһӣбҹҮ бһ бһҫбһҷбһңбһ¶бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”  
- бһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҖбһҹбһ¶бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бҹ”  

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶  

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–  

- бһҖбһҹбһ¶бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҳбһҪбһҷбҹ”  
- бһҖбҹҶбһҺбһҸбҹӢбһҠбҹӮбһ“бһҖбҹҶбһҺбһҸбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ”бҹ’бһҡбһҳбһјбһӣбһ•бҹ’бһҸбһ»бҹҶ (meta prompts)бҹ”  
- бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ DALL-E бһ“бһ·бһ„ Midjourneyбҹ”  

## бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһ”бһ¶бһ“бһҮбһ¶бһҖбһҹбһ¶бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–?  

бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһӮбһәбһҮбһ¶бһңбһ·бһ’бһёбһӣбҹ’бһўбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһ–бһёбһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ“бҹғGenerative AIбҹ” бһңбһ¶бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбҹ–  

- **бһҖбҹӮбһ”бҹ’бһҡбҹӮ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҶбһҡбһјбһ”бһ—бһ¶бһ–**бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ бһҠбһјбһ…бһҮбһ¶бһҖбһ¶бһҡбһҖбҹӮбһ”бҹ’бһҡбҹӮбһҡбһјбһ”бһ—бһ¶бһ– бһ“бһ·бһ„бһҖбһ¶бһҡбһ•бҹ’бһҹбҹҶбһҡбһјбһ”бһ—бһ¶бһ–бҹ”  

- **бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбҹ’бһ“бһ»бһ„бһңбһ·бһҹбҹҗбһҷбһ•бҹ’бһҹбҹҒбһ„бҹ—**бҹ” бһңбһ¶бһҖбҹҸбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһңбһ·бһҹбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ бһңбһ·бһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҮбһёбһңбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡ бһ‘бҹҒбһҹбһ…бһҡбһҺбҹҚ бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ бҹ’бһӮбҹҒбһҳ бһ“бһ·бһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбҹ”  

## бһҹбҹҒбһ…бһҖбҹ’бһҠбһёбһҹбһ“бҹ’бһ“бһ·бһҠбҹ’бһӢбһ¶бһ“бҹ– Edu4All  

бһҮбһ¶бһ•бҹ’бһ“бҹӮбһҖбһҳбһҪбһҷбһ“бҹғбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһҷбһҫбһ„бһ“бһ№бһ„бһ”бһ“бҹ’бһҸбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҹбҹ’бһҸбһ¶бһҸбһўбһ¶бһ”бҹӢбһҡбһ”бһҹбҹӢбһҷбһҫбһ„ Edu4All бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбҹ” бһҳбһ“бҹ’бһҡбҹ’бһҸбһёбһҡбҹҖбһ“бһ“бһ№бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһңбһ¶бһҷбһҸбһҳбҹ’бһӣбҹғбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒ бһўбҹ’бһңбһёбһҠбҹӮбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһ“бҹ„бҹҮбһӮбһәбһ‘бҹ…бһҸбһ¶бһҳбһ…бҹҶбһҺбһ„бҹӢбһ…бҹҶбһҺбһјбһӣбһ…бһ·бһҸбҹ’бһҸбһҡбһ”бһҹбҹӢбһҹбһ·бһҹбҹ’бһҹ бһ”бҹүбһ»бһ“бҹ’бһҸбһ¶бһ–бһҪбһҖбһӮбҹҒбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ”бҹ’бһҡбһҠбһ¶бһ”бҹӢбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһ•бҹӮбһ“бһҠбһёбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒ бһ¬бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбһҪбһўбһ„бҹ’бһӮбһҗбҹ’бһҳбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒ бһ¬бһҮбһҪбһҷбһ–бһҪбһҖбһӮбҹҒбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҳбһ¶бһ“бһ—бһ¶бһ–бһғбһҫбһүбһ…бҹ’бһ”бһ¶бһҹбҹӢбһӮбҹҶбһ“бһ·бһҸ бһ“бһ·бһ„бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҡбһ”бһҹбҹӢбһ–бһҪбһҖбһӮбҹҒбҹ”

бһҮбһ¶бһҳбһ»бһ“бһҹбһ·бһ“ [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) бһӮбһәбһҮбһ¶бһӮбҹҶбһҡбһј Generative AI бһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҹбҹҶбһҺбһ»бҹҶбһҡбһ¶бһ„бһҸбҹ’бһҡбһ„бҹӢ (transformer architecture) бһҮбһ¶бһҳбһҪбһҷбһ“бһ№бһ„ *autoregressive transformer*бҹ”  

*autoregressive transformer* бһҖбҹҶбһҺбһҸбҹӢбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһӮбҹҶбһҡбһјбһ”бһ„бҹ’бһҖбһҫбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһ–бһ·бһ–бһҺбҹҢбһ“бһ¶бһўбһҸбҹ’бһҗбһ”бһ‘ бһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбһӮбҹ’бһҡбһ¶бһ”бҹӢбһ–бһ·бһ… (pixel) бһҳбһҪбһҷбһ–бҹҒбһӣ бһ бһҫбһҷбһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһ”бҹ’бһҡбһҫбһӮбҹ’бһҡбһ¶бһ”бҹӢбһ–бһ·бһ…бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһӮбҹ’бһҡбһ¶бһ”бҹӢбһ–бһ·бһ…бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбҹ” бһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ…бҹ’бһҡбҹ„бҹҮбһҸбһ¶бһҳбһҹбҹ’бһҡбһ‘бһ¶бһ”бҹӢбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ”бһҺбҹ’бһҠбһ¶бһүбһ”бҹ’бһҡбһҹбһ·бһҸбһҡбһ бһјбһҸбһҡбһјбһ”бһ—бһ¶бһ–бһ”бһүбҹ’бһ…бһ”бҹӢбҹ”  

бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ“бҹҒбҹҮ DALL-E бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһӣбһҖбҹ’бһҒбһҺбҹҮ, бһңбһҸбҹ’бһҗбһ», бһӣбһҖбҹ’бһҒбһҺбҹҲбһ–бһ·бһҹбҹҒбһҹ, бһ“бһ·бһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ‘бҹҖбһҸбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһңбһ¶бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ‘бҹ„бҹҮбһҮбһ¶бһҷбҹүбһ¶бһ„бһҺбһ¶ DALL-E 2 бһ“бһ·бһ„ 3 бһҳбһ¶бһ“бһҖбһ¶бһҡбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ„бһӣбһҫбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ¶бһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһҳбһ»бһ“бҹ”

    ```python
        image_dir = os.path.join(os.curdir, 'images')

        # бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҗбһҸбһҳбһ·бһ“бһҳбһ¶бһ“, бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһ¶
        if not os.path.isdir(image_dir):

        image_path = os.path.join(image_dir, 'generated-image.png')

        # бһ‘бһ¶бһүбһҷбһҖбһҡбһјбһ”бһ—бһ¶бһ–бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸ
        print(generation_response)

        image_url = generation_response.data[0].url # бһҠбһҖ URL бһҡбһјбһ”бһ—бһ¶бһ–бһ–бһёбһҖбһ¶бһҡбһҶбҹ’бһӣбһҫбһҷбһҸбһ”
        generated_image = requests.get(image_url).content  # бһ‘бһ¶бһүбһҷбһҖбһҡбһјбһ”бһ—бһ¶бһ–
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # бһ”бһ„бҹ’бһ бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҫбһӣбһҡбһјбһ”бһ—бһ¶бһ–бһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ
        image = Image.open(image_path)
        image.show()

    # бһ…бһ¶бһ”бҹӢбһҖбһҡбһҺбһёбһҖбҹҶбһ бһ»бһҹ
    except openai.BadRequestError as err:
        print(err)

    ```
