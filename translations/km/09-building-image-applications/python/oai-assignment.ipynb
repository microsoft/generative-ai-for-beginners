> [DALL-E ជាការបញ្ចូលគ្នារវាងម៉ូដែលពីរនាក់ គឺ CLIP និង diffused attention](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst)។  

- **CLIP** ជាម៉ូដែលបង្កើត embeddings ដែលជាចំណាត់ថ្នាក់លេខសម្រាប់ទិន្នន័យ ពីរូបភាពនិងអត្ថបទ។

*Autoregressive transformer* កំណត់របៀបដែលម៉ូដែលបង្កើតរូបភាពពីការពិពណ៌នាអត្ថបទវាបង្កើត pixel មួយក្នុងមួយពេល ហើយបន្ទាប់មកប្រើ pixel នោះក្នុងការបង្កើត pixel បន្ទាប់។ វាត្រូវបានផ្លាស់ប្តូរជាលំដាប់តាមស្រទាប់ច្រើននៅក្នុងបណ្តាញឆ្លាទ្វារដោយដំណើរការដល់ពេលរូបភាពបញ្ចប់។

ជាមួយនឹងដំណើរការនេះ DALL-E អាចគ្រប់គ្រងលក្ខណៈ ពិសេស ផលិតផល និងលក្ខណៈផ្សេងទៀតក្នុងរូបភាពដែលវាបង្កើតឡើង។ ទោះយ៉ាងណា DALL-E 2 និង 3 មានការគ្រប់គ្រងកាន់តែច្រើនលើរូបភាពដែលបានបង្កើត,


## ការបង្កើតកម្មវិធីបង្កើតរូបភាពដំបូងរបស់អ្នក

តើអ្វីខ្លះដែលត្រូវការដើម្បីបង្កើតកម្មវិធីបង្កើតរូបភាព? អ្នកត្រូវការញឹកញាប់បណ្ណាល័យដូចខាងក្រោម៖

- **python-dotenv**, យើងណែនាំយ៉ាងខ្លាំងឱ្យអ្នកប្រើបណ្ណាល័យនេះដើម្បីរក្សាសម្ងាត់របស់អ្នកនៅក្នុងឯកសារ *.env* ឆ្ងាយពីកូដ។
- **openai**, បណ្ណាល័យនេះគឺជាអ្វីដែលអ្នកនឹងប្រើដើម្បីធ្វើប្រតិបត្តិការជាមួយ API OpenAI។
- **pillow**, ដើម្បីធ្វើការ​ជាមួយរូបភាពក្នុង Python។
- **requests**, ដើម្បីជួយអ្នកធ្វើការស្នើរសុំ HTTP។

1. បង្កើតឯកសារ *.env* ជាមួយមាតិកាដូចខាងក្រោម៖

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```


1. ប្រមាញ់បណ្ណាល័យខាងលើទាំងអស់ក្នុងឯកសារដែលមានឈ្មោះ *requirements.txt* ដូចខាងក្រោម៖

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. បន្ទាប់មក បង្កើតបរិយាកាសសិតិវិធីនិងដំឡើងបណ្ណាល័យទាំងនោះ៖


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. បញ្ចូលកូដខាងក្រោមក្នុងឯកសារដែលមានឈ្មោះ *app.py*:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # នាំចូល dotenv
    dotenv.load_dotenv()

    # បង្កើតវត្ថុ OpenAI
    client = OpenAI()


    try:
        # បង្កើតរូបភាពដោយប្រើ API បង្កើតរូបភាព
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # បញ្ចូលអត្ថបទពាក្យបញ្ជារបស់អ្នកនៅទីនេះ
            size='1024x1024',
            n=1
        )
        # កំណត់ថតសម្រាប់រូបភាពដែលបានរក្សាទុក
        image_dir = os.path.join(os.curdir, 'images')

        # ប្រសិនបើថតមិនមានបង្កើតវា
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ផ្ដើមផ្លូវរូបភាព (ចំណាំប្រភេទឯកសារ គួរតែជា png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # ទាញយករូបភាពដែលបានបង្កើត
        print(generation_response)

        image_url = generation_response.data[0].url # ដកយក URL រូបភាពពីការឆ្លើយតប
        generated_image = requests.get(image_url).content  # ទាញយករូបភាព
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # បង្ហាញរូបភាពក្នុងកម្មវិធីមើលរូបភាពលំនាំដើម
        image = Image.open(image_path)
        image.show()

    # จับករណីកំហុស
    except client.error.InvalidRequestError as err:
        print(err)

    ```

### លម្អិតបន្ថែមអំពីការបង្កើតរូបភាព

មកមើលកូដដែលបង្កើតរូបភាពលម្អិតជាងនេះ៖

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # បញ្ចូលអត្ថបទបញ្ជានៅទីនេះ
        size='1024x1024',
        n=1
    )
```

- **prompt** គឺជាចំណាំអត្ថបទដែលប្រើសម្រាប់បង្កើតរូបភាព។ ក្នុងករណីនេះ យើងកំពុងប្រើ prompt "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils"។
- **size** គឺជាបរិមាណទំហំរូបភាពដែលបានបង្កើត។ ក្នុងករណីនេះ យើងកំពុងបង្កើតរូបភាពដែលមានទំហំ 1024x1024 ពិក្សែល។
- **n** គឺជាចំនួនរូបភាពដែលបានបង្កើត។ ក្នុងករណីនេះ យើងកំពុងបង្កើតរូបពីរ។

មានអ្វីយ៉ាងផ្សេងទៀតដែលអ្នកអាចធ្វើជាមួយរូបភាពដែលយើងនឹងគ្របក្រោមនៅផ្នែកក្រោយ។

## សមត្ថភាពបន្ថែមនៃការបង្កើតរូបភាព

អ្នកបានឃើញរហូតមកថា តើយើងអាចបង្កើតរូបភាពដោយប្រើបន្ទាត់កូដព្រឹក Python ហើយ។ ទោះជាយ៉ាងណា មានអ្វីផ្សេងទៀតដែលអ្នកអាចធ្វើជាមួយរូបភាព។

អ្នកក៏អាចធ្វើបានដូចតទៅ៖

- **អនុវត្តកែប្រែ**។ ដោយផ្តល់រូបភាពមានស្រាប់ ម្សៅ​អាង (mask) និង prompt ត្រូវបានផ្តល់ អ្នកអាចកែប្រែរូបភាព។ ឧទាហរណ៍ អ្នកអាចបន្ថែមអ្វីមួយទៅកន្លែងខ្លះនៃរូបភាព។ សូមនឹកស្រមៃរូបភាពទន្សាយរបស់យើង អ្នកអាចបន្ថែមមួកទៅលើទន្សាយ។ វាដូចអ្នកផ្តល់រូបភាព ម្សៅអាង (កំណត់ផ្នែកត្រូវប្ដូរ) និងអត្ថបទ prompt ដើម្បីប្រាប់ពីអ្វីដែលគួរត្រូវធ្វើ។

    ```python
    response = openai.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    រូបភាពមូលដ្ឋានគឺមានតែទន្សាយប៉ុណ្ណោះ ប៉ុន្តែរូបភាពចុងក្រោយនឹងមានមួកលើទន្សាយ។
    
- **បង្កើតចម្លងផ្សេងៗ**។ គំនិតគឺអ្នកយករូបភាពមានស្រាប់ ហើយស្នើឲ្យបង្កើតចម្លងផ្សេងៗ។ ដើម្បីបង្កើតចម្លង អ្នកផ្តល់រូបភាព និងចំណាំអត្ថបទ និងកូដដូចនេះ៖

    ```python
    response = openai.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
