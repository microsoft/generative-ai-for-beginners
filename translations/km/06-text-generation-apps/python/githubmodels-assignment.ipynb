# бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһ‘бһӣбҹӢбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһҖбҹ’бһ“бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбһҗбһ¶бһҳбһ¶бһ“бһҖбҹҶбһ“бһҸбҹӢбһҹбһ¶бһҡвҖӢбһ’бҹҶбҹ—бһҠбһјбһ…бһҮбһ¶бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶ бһ“бһ·бһ„бһҳбһҪбһҷбһңбһ·бһҹбҹҗбһҷбһҳбһҪбһҷбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ В«бһңбһ·бһҹбҹ’бһңбһҖбһҳбҹ’бһҳбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶В»бҹ” бһӮбҹ’бһҡбһҝбһ„бһ…бһҖбҹ’бһҡбһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһўбҹ’бһ“бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹҶбһ•бһ»бһҹбһўбҹ’бһңбһёбһҳбһҪбһҷбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһўбҹ’бһ“бһҖбһ”бһүбҹ’бһ…бһјбһӣбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҠбһјбһ…бһ“бҹҒбҹҮбһ…бһјбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶, бһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһ”бҹӢ бһ“бһ·бһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһҮбһ¶бһ–бһ·бһҸбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ

бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бҹ–

- бһҹбҹ’бһңбҹӮбһ„бһҷбһӣбҹӢбһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбҹҶбһ“бһ·бһҸбһҹбҹ’бһ“бһјбһӣбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҷбһӣбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶, бһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһ“бһ·бһ„бһҹбһүбҹ’бһүбһ¶бһ‘бҹҶбһ бҹҶбһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҡбҹҖбһ“

бһ“бҹ…бһ…бһ»бһ„бһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶ бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҫбһҹбһүбҹ’бһүбһ¶бһ‘бҹҶбһ бҹҶбһ…бҹ’бһҡбһҫбһ“ бһ¬бһҸбһ·бһ… бһ бһҫбһҷбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҠбһјбһҡбһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһӣбһ‘бҹ’бһ’бһ•бһӣбһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бҹ”

## бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһё?

бһ’бһҳбҹ’бһҳбһҸбһ¶ бһ“бҹ…бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһҳбҹ’бһҠбһ„бһҺбһ¶бһҳбһҪбһҷбһңбһ¶бһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ№бһҖбҹ’бһҹбһ¶бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ—бҹҒбһ‘ console бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһ“бһ№бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһҳбһҪбһҷбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бҹ”
- бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ№бһҖбҹ’бһҹбһ¶бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ(UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ№бһҖбҹ’бһҹбһ¶бһ”бһ„бҹ’бһ бһ¶бһүбһҡбһјбһ”бһ—бһ¶бһ–(GUI) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһңбһ¶бһҷбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бҹ–

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһҺбһ¶бһҳбһҪбһҷбһ”бһ¶бһ“бһ‘бҹҒ бһӮбҹ’бһҡбһ¶бһ“бҹӢбһҸбҹӮбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”
- **бһҮбһ¶бһ—бһ¶бһҹбһ¶бһ–бһ·бһҹбҹҒбһҹ**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҮбһ¶бһ—бһ¶бһҹбһ¶бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҠбҹ„бһҷбһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҹбһ¶бһ„бһҹбһ„бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһҳбһҪбһҷ бһ‘бҹ„бҹҮбһ”бһёбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ”бһ“бҹ’бһҗбҹӮбһҳбҹ”

бһҳбһ¶бһ“бһҮбһҳбҹ’бһҡбһҫбһҹбһ–бһёбһҡбһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮбҹ–

### 1.Codespaces 

* бһ–бһёбһ•бҹ’бһ‘бһ¶бҹҶбһ„ 'Get started' бһҮбҹ’бһҡбһҫбһҹ "Run codespace"
* бһ”бһ„бҹ’бһҖбһҫбһҸ codespace бһҗбҹ’бһҳбһё(бһ¬бһ”бҹ’бһҡбһҫ codespace бһҳбһ¶бһ“бһҡбһҪбһ…бһҳбһҪбһҷ)
* VS Code бһ“бһ№бһ„бһ”бһҫбһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһЎбҹ…бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһ”бҹ’бһҡбһўбһ”бҹӢбһҠбҹ„бһҷбһҳбһ¶бһ“бһҹбҹҖбһңбһ—бҹ…бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһӣбҹҶбһўбһ·бһҸбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„

### 2. VS Code (бһ¬ IDE бһҠбҹӮбһӣбһ…бһјбһӣбһ…бһ·бһҸбҹ’бһҸ)

бһ–бһёбһ”бҹҠбһјбһҸбһ»бһ„бһ–бһҺбҹҢбһ”бҹғбһҸбһ„ 'Get started' бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ–бҹҗбһҸбҹҢбһҳбҹүбҹҚбһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ бһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҸбҹӢбһҖбҹ’бһ“бһ»бһ„ IDE бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһјбһӣбһ…бһ·бһҸбҹ’бһҸбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ“бҹҒбҹҮбһ”бһ„бҹ’бһ бһ¶бһү VS Code

* бһҮбҹ’бһҡбһҫбһҹбһ—бһ¶бһҹбһ¶ бһ“бһ·бһ„ SDK - бһҖбҹ’бһ“бһ»бһ„бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ“бҹҒбҹҮбһҮбҹ’бһҡбһҫбһҹ Python бһ“бһ·бһ„ Azure AI Inference SDK
* бһ”бһ„бҹ’бһҖбһҫбһҸ personal access token бһ“бҹ… GitHubбҹ” бһңбһ¶бһ“бһ№бһ„бһҹбҹ’бһҗбһ·бһҸбһҖбҹ’бһ“бһ»бһ„бһ•бҹ’бһ“бҹӮбһҖ Developer Settingsбҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһ•бҹ’бһҠбһӣбҹӢбһҹбһ·бһ‘бҹ’бһ’бһ·бһҺбһ¶бһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ token бһ“бҹ„бҹҮбһ‘бҹҒбҹ” бһ…бҹҶбһҺбһ¶бҹҶбһҗбһ¶ token бһ“бһ№бһ„бһ•бҹ’бһүбһҫбһ‘бҹ…бһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ Microsoftбҹ”
* бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҗбҹҒбһҡбһ”бһҡбһ·бһҹбҹ’бһҗбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҖбҹ’бһҹбһ¶ personal access token бһҡбһ”бһҹбҹӢ GitHub - бһҳбһ¶бһ“бһӮбҹҶбһҡбһјбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ bash, powershell бһ“бһ·бһ„ command prompt Windows 
* бһҠбҹҶбһЎбһҫбһ„бһҖбһ¶бһҡбһ–бһ№бһ„бһ•бҹ’бһўбҹӮбһҖбҹ– ```pip install azure-ai-inference```
* бһ…бһҳбҹ’бһӣбһ„бһҖбһјбһҠбһӮбҹҶбһҡбһјбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ‘бҹ…бһҜбһҖбһҹбһ¶бһҡ .py бһҳбһҪбһҷ
* бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҖбһ“бҹ’бһӣбҹӮбһ„бһҠбҹӮбһӣбһҡбһҖбҹ’бһҹбһ¶бһҖбһјбһҠ бһ бһҫбһҷбһҡбһҸбҹӢбһҜбһҖбһҹбһ¶бһҡбҹ– ```python filename.py```

бһҖбһ»бҹҶбһ—бҹ’бһӣбҹҒбһ… бһҗбһ¶ бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Azure AI Inference SDK бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҳбҹүбһјбһҠбҹӮбһӣбһ•бҹ’бһҹбҹҒбһ„бҹ—бһҠбҹ„бһҷбһҖбһ¶бһҡбһ•бҹ’бһӣбһ¶бһҹбһ”бҹҶбһӣбҹӮбһ„бһҗбҹ’бһӣбҹғ `model_name` бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбҹ”

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
