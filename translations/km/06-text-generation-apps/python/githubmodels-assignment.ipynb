# បង្កើតកម្មវិធីបង្កើតអត្ថបទ

តាំងពីពេលនេះ អ្នកបានឃើញថាមានគំនិតមូលដ្ឋានដូចជា prompts និងក៏មានវិស័យមួយទូទាំងហៅថា "prompt engineering"។ ឧបករណ៍ជាច្រើនដែលអ្នកអាចធ្វើការជាមួយដូចជា ChatGPT, Office 365, Microsoft Power Platform និងផ្សេងទៀត គាំទ្រឱ្យអ្នកប្រើប្រាស់ prompts ដើម្បីការបំពេញមួយចំនួន។

ដើម្បីឱ្យអ្នកបន្ថែមបទពិសោធន៍បែបនេះទៅកម្មវិធីមួយ អ្នកត្រូវយល់ដឹងពីកុងសែបដូចជា prompts, completions ហើយជ្រើសគ្រឿងបណ្ណាល័យមួយដើម្បីធ្វើការជាមួយ។ នេះជាអ្វីដែលអ្នកនឹងរៀននៅក្នុងជំពូកនេះ។

## ការណែនាំ

ក្នុងជំពូកនេះ អ្នកនឹងៈ

- រៀនអំពីបណ្ណាល័យ openai និងគំនិតមូលដ្ឋានរបស់វា។
- បង្កើតកម្មវិធីបង្កើតអត្ថបទប្រើប្រាស់ openai។
- យល់ពីរបៀបប្រើប្រាស់គំនិតដូចជា prompt, temperature និង tokens ដើម្បីបង្កើតកម្មវិធីបង្កើតអត្ថបទ។

## គោលដៅរៀន

នៅចុងម៉ោងសិក្សានេះ អ្នកនឹងអាច៖

- លើក​យក​ថាតើកម្មវិធី​បង្កើត​អត្ថបទ​ជាអ្វី។
- បង្កើតកម្មវិធីបង្កើតអត្ថបទប្រើប្រាស់ openai។
- កំណត់កម្មវិធីរបស់អ្នកឱ្យប្រើប្រាស់ tokens ច្រើន ឬតិច និងប្ដូរ temperature ដើម្បីទទួលបានលទ្ធផលផ្សេងៗគ្នា។

## កម្មវិធីបង្កើតអត្ថបទជាអ្វី?

នៅពេលដែលអ្នកបង្កើតកម្មវិធី វាមានមុខងារផ្សេងៗដូចជា៖

-Command-based។ កម្មវិធី console ជាកម្មវិធីដែលអ្នកវាយបញ្ជា ហើយវានឹងអនុវត្តភារៈកិច្ចមួយ។ ឧទាហរណ៍ `git` គឺជាកម្មវិធីដំណើរការតាមបញ្ជា។
-User interface (UI)។ មានកម្មវិធីមួយចំនួនមាន graphical user interfaces (GUIs) ដែលអ្នកចុចប៊ូតុង វាយអត្ថបទ ជ្រើសរើសជម្រើស និងផ្សេងៗទៀត។

### កម្មវិធី Console និង UI មានកំណត់

ប្រៀបធៀបនឹងកម្មវិធី command-based ដែលអ្នកវាយបញ្ជា៖

- **វាមានកំណត់**។ អ្នកមិនអាចវាយបញ្ជាណាមួយក៏បានទេ មានតែបញ្ជាដែលកម្មវិធីគាំទ្រតែប៉ុណ្ណោះ។
- **ភាសាផ្ដល់ជូន**។ កម្មវិធីខ្លះគាំទ្រភាសាច្រើន ប៉ុន្តែលំនាំដើមកម្មវិធីត្រូវបានបង្កើតសម្រាប់ភាសាមួយ ជាប្រភេទមួយ ទោះបីអ្នកអាចបញ្ចូលសារពើភ័ណ្ឌភាសាបន្ថែម។

### អត្ថប្រយោជន៍នៃកម្មវិធីបង្កើតអត្ថបទ

តើកម្មវិធីបង្កើតអត្ថបទខុសគ្នាយ៉ាងដូចម្តេច?

ក្នុងកម្មវិធីបង្កើតអត្ថបទ អ្នកមានភាពបត់បែនច្រើន អ្នកមិនបានកំណត់ត្រឹមបញ្ជាឬភាសាកំណត់។ ជំនួសវិញ អ្នកអាចប្រើភាសាធម្មជាតិក្នុងការ tương tácជាមួយកម្មវិធី។ អត្ថប្រយោជន៍មួយទៀតគឺដោយសារថាអ្នកកំពុងប្រើប្រភពទិន្នន័យដែលបានហ្វឹកហាត់លើបណ្ដុំព័ត៌មានធំមួយ ខុសពីកម្មវិធីលំនាំដើមដែលពិបាកកំណត់ត្រឹមតែអ្វីដែលមាននៅក្នុងបណ្តុំព័ត៌មាន។

### ខ្ញុំអាចបង្កើតអ្វីជាមួយកម្មវិធីបង្កើតអត្ថបទ?

មានអ្វីជាច្រើនដែលអ្នកអាចបង្កើតបាន។ ឧទាហរណ៍៖

- **Chatbot**។ Chatbot ឆ្លើយសំនួរអំពីប្រធានបទជាច្រើន ដូចជាក្រុមហ៊ុននិងផលិតផលរបស់អ្នកអាចជាជម្រើសល្អ។
- **ជំនួយការ**។ LLMs អាចធ្វើបានល្អក្នុងការសង្ខេបអត្ថបទ ទទួលយកព័ត៌មានពីអត្ថបទ បង្កើតអត្ថបទ ដូចជា resume និងផ្សេងៗទៀត។
- **ជំនួយគោដៅកូដ**។ ដោយផ្អែកលើម៉ូឌែលភាសាដែលអ្នកប្រើ អ្នកអាចបង្កើតជំនួយអំពីកូដដែលជួយអ្នកសរសេរកូដ។ ឧទាហរណ៍ អ្នកអាចប្រើផលិតផលដូចជា GitHub Copilot និង ChatGPT ដើម្បីជួយសរសេរកូដ។

## តើខ្ញុំអាចចាប់ផ្តើមដោយរបៀបណា?

អ្នកត្រូវស្វែងរកវិធីសាស្រ្តបញ្ចូលជាមួយ LLM ដែលភាគច្រើនរួមមានវិធីសាស្រ្តពីរនេះ៖

- ប្រើ API។ នៅទីនេះអ្នកបង្កើតសំណើវែបជាមួយ prompt របស់អ្នក ហើយទទួលអត្ថបទដែលបានបង្កើតតបឡេីងវិញ។
- ប្រើបណ្ណាល័យ។ បណ្ណាល័យជួយបិទបញ្ជី API ហើយធ្វើឱ្យវាងាយស្រួលប្រើ។

## បណ្ណាល័យ/SDKs

មានបណ្ណាល័យជាច្រើនដែលល្បីសម្រាប់ធ្វើការជាមួយ LLMs ដូចជា៖

- **openai** បណ្ណាល័យនេះធ្វើឱ្យងាយស្រួលក្នុងការតភ្ជាប់ទៅម៉ូឌែលរបស់អ្នក និងផ្ញើ prompts ។

បន្ទាប់ពីនេះ មានបណ្ណាល័យដែលដំណើរការលើកម្រិតខ្ពស់ជាង៖

- **Langchain**។ Langchain ល្បី ហើយគាំទ្រភាសា Python។
- **Semantic Kernel**। Semantic Kernel ជាបណ្ណាល័យពី Microsoft គាំទ្រភាសា C#, Python និង Java។

## អនុវត្ត - បង្កើតមុខម្ហូបអាហារមួយ

ឥឡូវនេះដែលយើងបានបង្ហាញសេណារីយ៉ូមួយរួចរាល់ នាំឱ្យយើងសរសេរកូដឲ្យស្របទៅនឹងសេណារីយ៉ូដែលបានបង្ហាញ។ ដើម្បីធ្វើដូចនេះ សូមធ្វើតាមជំហានខាងក្រោម៖

1. ប្រើឯកសារដែលមានរួចជាថ្នាំងចាប់ផ្តើម
1. បង្កើតអថេរ `prompt` ហើយផ្លាស់ប្តូរកូដឧទាហរណ៍ដូចខាងក្រោម៖


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
