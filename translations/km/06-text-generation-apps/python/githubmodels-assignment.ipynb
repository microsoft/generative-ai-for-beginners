# бһҹбҹҶбһҺбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһҘбһЎбһјбһңбһ“бҹ…бһҸбһ¶бһҳбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮбһҗбһ¶ бһҳбһ¶бһ“бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһӮбҹ’бһҡбһ№бҹҮбһҠбһјбһ…бһҮбһ¶бһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡ бһ“бһ·бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһ“бһјбһңбһңбһ·бһҹбҹҗбһҷбһҳбһҪбһҷбһҳбһҪбһҷбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ "бһңбһ·бһҹбҹ’бһңбһҖбһҳбҹ’бһҳбһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡ"бҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһўбһ“бҹ’бһҸбһҡбһ¶бһӮбһҳбһ“бҹҚбһ”бһ¶бһ“бһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ•бҹ’бһӣбһјбһңбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһұбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһҺбһ¶бһҳбһҪбһҷбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһўбҹ„бһҷбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһҳбһ“бһ»бһҹбҹ’бһҹбһҸбҹ’бһҡбһјбһңбһҸбҹӮбһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ бһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡ, бһ”бһ“бҹ’бһ‘бһ”бҹӢбһ”бһүбҹ’бһ…бһ”бҹӢ бһ“бһ·бһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһҳбҹ’бһҡбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ

бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„:

- бһҡбҹҖбһ“бһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбһ“бҹ’бһӣбһ„бһӮбҹ’бһҡбһ№бҹҮбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһӮбһ“бҹ’бһӣбһ„бһҒбҹ’бһӣбҹҮбҹ—бһҠбһјбһ…бһҮбһ¶бһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡ, бһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһ“бһ·бһ„бһҸбһјбһҒбҹҒбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҡбҹҖбһ“

бһ“бҹ…бһ…бһ»бһ„бһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…:

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһӮбһәбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҸбһјбһҒбҹҒбһ“ бһ“бһ·бһ„бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ¶бһҡбһ…бҹҒбһүбһ•бҹ’бһҹбһ¶бһҷбһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бҹ”

## бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһҮбһ¶бһўбҹ’бһңбһё?

бһҮбһ¶бһ‘бһјбһ‘бҹ…бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһңбһ¶бһ“бһ№бһ„бһҳбһ¶бһ“бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһҸбҹҶбһҺбһ¶бһ„бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҠбһјбһ…бһҸбһ‘бҹ…бҹ–

- бһӮбҹҶбһҡбһјбһ”бһүбҹ’бһҮбһ¶бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһҳбҹ’бһҡбһјбһ”бһүбҹ’бһҮбһ¶ бһӮбһәбһҮбһ¶вҖӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһўбһ“бһ»бһңбһҸбҹ’бһҸбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“бһҖбһ¶бһҡбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбҹҶбһҡбһјбһ”бһүбҹ’бһҮбһ¶бҹ”
- бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“бһ•бҹ’бһ‘бһ¶бҹҶбһ„бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһҖ (GUI) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбҹҶбһҡбһјбһ”бһүбҹ’бһҮбһ¶ бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбҹҶбһҡбһјбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бҹ– 

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһўбҹ’бһңбһёбһҖбҹҸбһ”бһ¶бһ“бһ‘бһ¶бһӣбҹӢбһҸбҹӮбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ‘бҹҒбҹ”
- **бһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһ—бһ¶бһҹбһ¶**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһҠбҹ„бһҷбһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳ бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһ—бһ¶бһҹбһ¶бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ”бһ¶бһ“бһ–бһёбһҖбҹ’бһҡбҹ„бһҷбҹ”

### бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҳбһ¶бһ“бһ—бһ¶бһ–бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбҹҒбһҡбһёбһ—бһ¶бһ–бһ…бҹ’бһҡбһҫбһ“ бһҮбһ¶бһ„бһҳбһ»бһ“бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһҖбҹҶбһҺбһҸбҹӢбһ“бҹ…бһӣбһҫбһ”бһүбҹ’бһҮбһ¶бһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ¬бһ—бһ¶бһҹбһ¶бһ”бһүбҹ’бһ…бһјбһӣбһҳбһҪбһҷбһ‘бҹҒбҹ” бһ”бҹ’бһҡбҹҖбһ”бһҡбһҪбһҳ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ”бҹ’бһҡбһ–бҹғбһҺбһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбһ“бҹ’бһҸбһҡбһ¶бһӮбһҳбһ“бҹҚбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбҹ” бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһҳбһҪбһҷбһ‘бҹҖбһҸбһӮбһәбһ–бһёбһ–бҹ’бһҡбҹ„бҹҮбһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһўбһ“бҹ’бһҸбһҡбһ¶бһӮбһҳбһ“бҹҚбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ бҹ’бһңбһ№бһҖбһ бһ¶бһҸбҹӢбһӣбһҫбһ”бһҺбҹ’бһҠбһ»бҹҶбһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ’бҹҶбһҳбһҪбһҷ бһҒбһ»бһҹбһ–бһёбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹғбһҺбһёбһҠбҹӮбһӣбһ…бҹҶбһ“бһ»бһ…бһ…бҹҒбһүбһ–бһёбһӮбҹ’бһҡбҹ„бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

### бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹ’бһңбһёбһ”бһ¶бһ“бһҒбҹ’бһӣбҹҮбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

- **бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбҹүбһјбһ”бһјбһҸбһҹбһ“бҹ’бһ‘бһ“бһ¶ (chatbot)**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбҹүбһјбһ”бһјбһҸбһҹбһ“бҹ’бһ‘бһ“бһ¶бһҠбҹӮбһӣбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһўбҹҶбһ–бһёбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘ бһҠбһјбһ…бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһңбһ¶ бһўбһ¶бһ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһӣбҹ’бһўбһҳбһҪбһҷбҹ”
- **бһҮбҹҶбһ“бһҪбһҷбһҖбһ¶бһҡ**бҹ” LLMs бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһўбһҹбҹ’бһ…бһ¶бһҡбҹ’бһҷбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘ бһҡбһҖбһҷбһӣбҹӢбһ–бһҸбҹҢбһҳбһ¶бһ“бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶бһ”бҹ’бһҡбһңбһҸбҹ’бһҸбһ·бһҡбһјбһ” бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”
- **бһҮбҹҶбһ“бһҪбһҷбһҖбһ¶бһҡбһҖбһјбһҠ**бҹ” бһҠбҹ„бһҷбһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбҹҶбһ“бһҪбһҷбһҖбһ¶бһҡбһҖбһјбһҠбһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһ“бһ·бһ„ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһўбһјбһҒбҹҒ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҸбҹӮбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ…бһјбһӣбһҡбһҪбһҳбһҮбһ¶бһҳбһҪбһҷ LLM бһҠбҹӮбһӣбһҮбһ¶бһ‘бһјбһ‘бҹ…бһҳбһ¶бһ“бһҡбһ”бҹҖбһ”бһ–бһёбһҡбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһңбҹүбһ·бһ”бһҮбһ¶бһҳбһҪбһҷбһ…бҹҶбһӣбһҫбһҷбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ‘бһ‘бһҪбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһ”бһ·бһҸбһ”бһүбҹ’бһҮбһ¶бһҖбҹӢбһ‘бҹ…бһӣбһҫбһҖбһ¶бһҡбһ бҹ… API бһ“бһ·бһ„бһ’бҹ’бһңбһҫбһұбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDK

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ–бҹҒбһүбһ“бһ·бһҷбһҳбһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ LLMs бһҠбһјбһ…бһҮбһ¶бҹ–

- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бҹ’бһүбһҫбһ…бҹҶбһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһ…бһјбһӣбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһӣбһҫбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҮбһ¶бһ„бһЎбһҫбһ„бҹ–

- **Langchain**. Langchain бһ–бҹҒбһүбһ“бһ·бһҷбһҳ бһ“бһ·бһ„бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ Pythonбҹ”
- **Semantic Kernel**. Semantic Kernel бһӮбһәбһҮбһ¶бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹ„бһҷ Microsoft бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ C#, Python бһ“бһ·бһ„ Javaбҹ”

бһ–бһёбһӮбһҳбҹ’бһҡбҹ„бһ„ Microsoft Foundry бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ‘бһ‘бһҪбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ¶бһ“бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҸбҹӢбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ IDE бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһјбһӣбһ…бһ·бһҸбҹ’бһҸбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ“бҹҒбҹҮбһ“бһ№бһ„бһ”бһ„бҹ’бһ бһ¶бһүбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ VS Code

* бһҸбҹҶбһЎбһҫбһ„бһҳбҹүбһјбһҠбҹӮбһӣ chat бһҠбһјбһ…бһҮбһ¶ `gpt-4o-mini` бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһӮбһҳбҹ’бһҡбҹ„бһ„ Foundry бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ
* бһ…бһҳбҹ’бһӣбһ„ **endpoint** бһ“бһ·бһ„ **API key** бһ–бһёбһ‘бҹҶбһ–бҹҗбһҡ Overview бһ“бҹғбһӮбһҳбҹ’бһҡбҹ„бһ„
* бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҗбҹҒбһҡбһ”бһҡбһ·бһҹбҹ’бһҗбһ¶бһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҖбҹ’бһҹбһ¶бһ‘бһ»бһҖ: `AZURE_INFERENCE_ENDPOINT` бһ“бһ·бһ„ `AZURE_INFERENCE_CREDENTIAL` - бһҳбһ¶бһ“бһӮбҹҶбһҡбһјбһҖбҹ’бһ“бһ»бһ„ bash, powershell бһ“бһ·бһ„ command prompt бһңбһёбһ“бһҠбһјбһҹ
* бһҠбҹҶбһЎбһҫбһ„бһўбҹ’бһңбһёбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ: ```pip install azure-ai-inference```
* бһ…бһҳбҹ’бһӣбһ„бһҖбһјбһҠбһӮбҹҶбһҡбһјбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ…бһјбһӣбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ .py
* бһ…бһјбһӣбһ‘бһёбһҸбһ¶бҹҶбһ„бһҠбҹӮбһӣбһҹбҹ’бһ“бһ¶бһҠбҹғбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҹбһ„бҹ’бһӮбҹ’бһҡбҹ„бҹҮ бһ бһҫбһҷбһҡбһҸбҹӢбһҜбһҖбһҹбһ¶бһҡ: ```python filename.py```

бһҖбһ»бҹҶбһ—бҹ’бһӣбҹҒбһ…бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ Azure AI Inference SDK бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһҖбһӣбҹ’бһ”бһ„бһҳбҹүбһјбһҠбҹӮбһӣбһ•бҹ’бһҹбҹҒбһ„бҹ—бһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһ„бһ¶бһҷ бһҮбһ¶бһҳбһҪбһҷбһҖбһ¶бһҡбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҸбһҳбҹ’бһӣбҹғ`model_name` бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбҹ”

бһҳбҹүбһјбһҠбҹӮбһӣбһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбһҳбһ¶бһ“бһҹбҹ’бһҡбһ¶бһ”бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҸбһ¶бһЎбһ»бһҖ Microsoft Foundry Models:

* OpenAI: gpt-4o, gpt-4o-mini, gpt-4.1, gpt-4.1-mini, o3-mini, text-embedding-3-large, text-embedding-3-small
* Meta: Meta-Llama-3.1-405B-Instruct, Meta-Llama-3.1-70B-Instruct, Meta-Llama-3.1-8B-Instruct, Llama-3.2-11B-Vision-Instruct, Llama-3.2-90B-Vision-Instruct, Llama-4-Scout-17B-16E-Instruct
* Mistral AI: Mistral-large-2411, Mistral-small-2503, Codestral-2501, Ministral-3B
* Microsoft: Phi-4, Phi-4-mini-instruct, Phi-4-multimodal-instruct, Phi-4-reasoning
* Cohere: Cohere-command-r-plus-08-2024, Cohere-embed-v3-multilingual, Cohere-embed-v3-english
* DeepSeek: DeepSeek-V3, DeepSeek-R1



**бһҘбһЎбһјбһңбһ‘бҹ…бһ•бҹ’бһ“бҹӮбһҖ 'Generate Text' бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ“бҹ’бһҸбһҖбһ¶бһҡбһ„бһ¶бһҡ**

## бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһҳбһҪбһҷ ChatCompletions

бһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһ”бҹ’бһҡбһҫбһҗбҹ’бһ“бһ¶бһҖбҹӢ `ChatCompletionsClient` бҹ”
бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„ `samples/python/azure_ai_inference/basic.py` бһ•бҹ’бһ“бҹӮбһҖбһҶбҹ’бһӣбһҫбһҷбһҸбһ”бһ“бҹғбһҖбһјбһҠ бһ”бһ“бҹ’бһҸбһ·бһҷбһјбһңбһ·бһ’бһё/бһҖбҹҶбһҺбһҸбҹӢбһҖбҹ’бһ“бһ»бһ„бһҳбһ»бһҒбһ„бһ¶бһҡ user role бһҠбҹ„бһҷбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһ–бһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡ content бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # бһ”бҹүбһ¶бһҡбҹүбһ¶бһҳбҹүбҹӮбһҸбҹ’бһҡвҖӢбһҠбҹӮбһӣбһҮбһҮбһҸбҹӢвҖӢбһўбһ¶бһ…вҖӢбһ…бһ¶бһ”бҹӢвҖӢбһ”бһ¶бһ“
    temperature=1.,
    max_tokens=1000,
    top_p=1.

```

бһҡбһҸбҹӢбһҜбһҖбһҹбһ¶бһҡбһҠбҹӮбһӣбһ”бһ¶бһ“бһўбһ¶бһ”бҹӢбһҠбҹҒбһҸбһҠбһҫбһҳбҹ’бһ”бһёбһҳбһҫбһӣбһӣбһ‘бҹ’бһ’бһ•бһӣ


## бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ“бҹғбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҠбһ№бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ— бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһ–бһёбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҠбһ№бһ„бҹ” бһўбҹ’бһ“бһҖз”ҡиҮібһҳбһ¶бһ“бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бҹ„бҹҮ бһ“бһ·бһ„бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҖбҹӮбһ”бҹ’бһҡбҹӮ бһ“бһ·бһ„бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бҹ’бһҡбһ—бҹҒбһ‘бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ”бһ¶бһ“бҹ” 

бһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҠбһ№бһ„бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһ•бҹ’бһҹбҹҒбһ„бҹ—бһӮбҹ’бһ“бһ¶бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

- **бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бҹ’бһҡбһ—бҹҒбһ‘бһҳбһҪбһҷ**бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбҹҶбһҺбһ¶бһ–бҹ’бһҷ, бһҹбҹҶбһҺбһҪбһҡбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ…бһјбһӣбһҡбһҪбһҳбһ”бҹ’бһҡбһЎбһ„ бһӣбҹ”
- **бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“**бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҠбһ№бһ„бһҠбһҫбһҳбҹ’бһ”бһёбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ 'бһҸбһҫ CORS бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһўбҹ’бһңбһёбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһўбһ—бһ·бһңбһҢбҹ’бһҚбһ”бһҺбҹ’бһҠбһ¶бһү?'бҹ”
- **бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһјбһҠ**бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһҖбһ¶бһҡбһ”бһ„бҹ’бһ бһ¶бһүбһІбҹ’бһҷбһҠбһ№бһ„бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһјбһҠ бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ”бһ„бҹ’бһҖбһҫбһҸбһ”бһҺбҹ’бһҠбһ¶бһүбһ”бһүбҹ’бһ…бҹҒбһүбһ‘бҹ’бһҡбһ„бҹӢбһ‘бҹ’бһҡбһ¶бһҷ бһҠбҹӮбһӣбһ”бҹ’бһҡбһҫбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ•бҹ’бһ‘бҹҖбһ„бһ•бҹ’бһ‘бһ¶бһҸбҹӢбһўбҹҠбһёбһҳбҹүбҹӮбһӣ бһ¬бһ бҹҒбһҸбһ»бһўбҹ’бһңбһёбһҳбһ·бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһ”бһ¶бһ“бһҠбһјбһ…бһҮбһ¶ бһңбҹҒбһ”бһҹбһ¶бһҷбһҳбһҪбһҷ?

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
