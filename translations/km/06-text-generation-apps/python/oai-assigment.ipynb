# бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһ‘бһӣбҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбһҸбһ¶бһҳбһҳбһ’бҹ’бһҷбҹ„бһ”бһ¶бһҷбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮбһҗбһ¶бһҳбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ prompt бһ бһҫбһҷбһҳбһ¶бһ“бһңбһ·бһҹбҹҗбһҷбһҳбһҪбһҷбһ‘бҹҖбһҸбһ бҹ…бһҗбһ¶ "prompt engineering"бҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҳбһҪбһҷбһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ prompt бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһўбҹ’бһңбһёбһҳбһҪбһҷбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҠбһјбһ…бһ“бҹҒбҹҮбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, completions бһ“бһ·бһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ

бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„:

- бһӣбһәбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҷбһӣбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, temperature бһ“бһ№бһ„ tokens бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҹбһ·бһҖбҹ’бһҹбһ¶

бһ“бҹ…бһ…бһ»бһ„аёҡаё—бһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…:

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһІбҹ’бһҷбһ”бҹ’бһҡбһҫ tokens бһ…бҹ’бһҡбһҫбһ“ бһ¬бһҖбҹҸбһҸбһ·бһ… бһ“бһ·бһ„бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡ temperature бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһӣбһ‘бҹ’бһ’бһ•бһӣбһ•бҹ’бһҹбҹҒбһ„бҹ—бһӮбҹ’бһ“бһ¶бҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһё?

бһҮбһ¶бһ’бһҳбҹ’бһҳбһҸбһ¶ бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһңбһ¶бһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ:

- Command-basedбҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ’бһҳбҹ’бһҳбһҸбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһ’бҹ’бһңбһҫбһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһҳбһҪбһҷбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё command-basedбҹ”
- бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһ“бһ№бһ„бһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһҖ (GUIs) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ”бһ“бҹ’бһҗбҹӮбһҳбһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһӮбҹ’бһҡбҹ„бҹҮбһҡбҹүбҹғбһҡбҹүбһ¶бһң

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһ‘бҹ…бһ“бһ№бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё command-based бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶:

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһҺбһ¶бһҳбһҪбһҷбһ”бһ¶бһ“бһ‘бҹҒ бһҳбһ¶бһ“бһҸбҹӮбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбҹ”
- **бһҮбһ¶бһ—бһ¶бһҹбһ¶бһ–бһ·бһҹбҹҒбһҹ**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҠбҹ„бһҷбһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҺбһ¶бһҳбһҪбһҷ бһҠбҹ„бһҷбһ”бһҫбһўбҹ’бһ“бһҖбһ…бһ„бҹӢбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ•бҹ’бһҹбҹҒбһ„бҹ”

### бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһ”бһҫбһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһ“бһ№бһ„бһҖбһ¶бһҡбһҒбҹ’бһӣбҹҮбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ—бһ¶бһ–бһ”бһҸбҹӢбһ”бҹӮбһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„ бһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһ…бһ¶бһҖбҹӢбһҖбҹҶбһҺбһҸбҹӢбһҠбҹ„бһҷбһ–бһ¶бһҖбҹ’бһҷбһ”бһүбҹ’бһҮбһ¶бһ¬бһ—бһ¶бһҹбһ¶бһ”бһүбҹ’бһ…бһјбһӣбһ–бһ·бһҹбҹҒбһҹбһ‘бҹҒбҹ” бһңбһ¶бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбҹ” бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһӮбһә бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһјбһӣбһ‘бҹ…бһ“бһ№бһ„бһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҳбһҪбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӣбһҫбһҳбһ¶бһҸбһ·бһҖбһ¶бһ’бҹҶбһ’бҹҒбһ„ бһҳбһ·бһ“бһҠбһјбһ…бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹғбһҺбһёбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһ‘бҹҒбҹ”

### бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһўбҹ’бһңбһёбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ…бҹ’бһ“бҹғбһ”бҹ’бһҡбһҢбһ·бһҸбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ:

- **бһҮбһ¶ chatbot**бҹ” Chatbot бһҠбҹӮбһӣбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһўбҹҶбһ–бһёбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘бһҠбһјбһ…бһҮбһ¶ бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһңбһ¶ бһўбһ¶бһ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһӣбҹ’бһўбҹ”
- **бһҮбһ¶бһҮбҹҶбһ“бһҪбһҷбһҖбһ¶бһҡ**бҹ” LLMs бһӣбҹ’бһўбһҖбҹ’бһ“бһ»бһ„бһҡбһҝбһ„бһҠбһјбһ…бһҮбһ¶ бһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘ бһ‘бһ¶бһүбһҷбһҖбһ…бҹҶбһҺбҹҒбҹҮбһҠбһ№бһ„бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ•бһӣбһ·бһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶ бһҡбһ”бһ¶бһҷбһҖбһ¶бһҡбһҺбҹҚбһҮбһ¶бһҠбһҫбһҳбҹ”
- **бһҮбһ¶бһҮбҹҶбһ“бһҪбһҷбһҖбһјбһҠ**бҹ” бһҡбһ¶бһ”бҹӢбһҸбһ¶бһҳбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһҮбҹҶбһ“бһҪбһҷбһҖбһјбһҠбһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһҡбһҪбһҳбһҮбһ¶бһҳбһҪбһҷ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһ…бҹ’бһ”бһ¶бһҹбҹӢбһҺбһ¶бһҹбҹӢ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһ·бһ’бһёбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҮбһ¶бһҳбһҪбһҷ LLM бһҠбҹӮбһӣбһ—бһ¶бһӮбһ…бҹ’бһҡбһҫбһ“бһӮбһәбһҳбһ¶бһ“бһ–бһёбһҡбһңбһ·бһ’бһёбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ:

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһ”бһҺбҹ’бһҠбһ¶бһүбһҮбһ¶бһҳбһҪбһҷ prompt бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ бһҫбһҷбһ‘бһ‘бһҪбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһ”бһ·бһҸбһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһ бҹ… API бһ бһҫбһҷбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDKs

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҠбҹӮбһӣбһӣбҹ’бһ”бһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ LLMs бһҠбһјбһ…бһҮбһ¶:

- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бҹ’бһүбһҫ promptбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹҒбҹҮбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһ”бҹ’бһҡбһҸбһ·бһ”бһҸбҹ’бһҠбһ·бһҖбһ¶бһҡбһ“бҹ…бһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҮбһ¶бһ„бһҠбһјбһ…бһҮбһ¶:

- **Langchain**. Langchain бһӣбҹ’бһ”бһёбһҒбҹ’бһӣбһ¶бҹҶбһ„ бһ“бһ·бһ„бһӮбһ¶бҹҶбһ‘бҹ’бһҡ Pythonбҹ”
- **Semantic Kernel**. Semantic Kernel бһҮбһ¶бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹ„бһҷ Microsoft бһҠбҹӮбһӣбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ C#, Python бһ“бһ·бһ„ Javaбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openai

бһ…бһ„бҹӢбһғбһҫбһүбһҗбһ¶ бһҸбһҫбһҷбһҫбһ„бһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ… бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҺбһ¶бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ бһ…бҹҶбһ“бһҪбһ“бһ”бҹүбһ»бһ“бҹ’бһҳбһ¶бһ“бһ“бһ№бһ„бһ’бҹ’бһңбһҫбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бҹ”

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. бһҳбһҖбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҖбһјбһҠбһҖбҹ’бһ“бһ»бһ„бһңбһ·бһ’бһёбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
