# бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһӣбһҫбһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ¶бһ“бҹҒбҹҮбһҗбһ¶бһҳбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҹбҹ’бһ“бһјбһӣбһҠбһјбһ…бһҮбһ¶ prompt бһ“бһ·бһ„бһҳбһ¶бһ“бһҸбҹҶбһ”бһ“бҹӢбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ вҖңбһ”бһ…бҹ’бһ…бҹҒбһҖбһңбһ·бһ‘бҹ’бһҷбһ¶бһ–бҹҒбһӣpromptвҖқ бһ•бһ„бһҠбҹӮбһҡбҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһ”бһ¶бһ“бһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһўбҹ’бһ“бһҖбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ prompt бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһҳбһҪбһҷбҹ”

бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбҹ’бһ“бһҖбһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҠбһјбһ…бһ“бҹҒбҹҮбһ‘бҹ…бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, completions бһ“бһ·бһ„бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһ”Intro

бһ“бҹ…бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„:

- бһҡбҹҖбһ“бһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбҹҶбһ“бһ·бһҸбһҹбҹ’бһ“бһјбһӣбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҷбһӣбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, temperature бһ“бһ·бһ„ tokens бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҡбҹҖбһ“

бһ“бҹ…бһ…бһ»бһ„бһҳбҹүбҹ„бһ„бһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…:

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҠбһҫбһҳбҹ’бһ”бһёбһ”бҹ’бһҡбһҫ tokens бһ…бҹ’бһҡбһҫбһ“ бһ¬бһҸбһ·бһ… бһ“бһ·бһ„бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡ temperature бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһӣбһ‘бҹ’бһ’бһ•бһӣбһҺбҹӮбһ“бһ”бҹ’бһҡбҹӮбһ”бҹ’бһҸбһјбһҡбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһё?

бһ’бһҳбҹ’бһҳбһҸбһ¶ бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһңбһ¶бһҳбһ¶бһ“бһҳбһ»бһҒбһ„бһ¶бһҡбһҡбһ¶бһ„бһҳбһҪбһҷбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶(Command-based)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһ’бһҳбҹ’бһҳбһҸбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһҹбҹҶбһҡбҹҒбһ…бһ“бһјбһңбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһҳбһҪбһҷбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶бҹ”
- бһҳбһ»бһҒбһ„бһ¶бһҡ user interface (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“ graphical user interfaces (GUIs) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё Console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶ бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бҹ–

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бһ“бҹ…бһӮбҹ’бһҡбһ”бҹӢбһ”бҹ’бһҡбһ—бҹҒбһ‘бһ”бһ¶бһ“бһ‘бҹҒ бһӮбһәбһҸбҹӮбһ“бҹ…бһӣбһҫбһ”бһүбҹ’бһҮбһ¶вҖӢбһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбҹ”
- **бһ—бһ¶бһҹбһ¶бһ…бҹ’бһ”бһ¶бһ”бҹӢбһ•бҹ’бһҸбһ¶бһ…бҹӢбһҳбһ»бһҒ**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҠбҹ„бһҷбһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҹбһ¶бһ„бһҹбһ„бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһ‘бҹ„бҹҮбһ”бһёбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбҹҸбһҠбҹ„бһҷбҹ”

### бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҳбһ¶бһ“бһ—бһ¶бһ–бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһ—бһ¶бһ–бһ”бһҸбҹӢбһ”бҹӮбһ“бһ…бҹ’бһҡбһҫбһ“бһҳбһҪбһҷ бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҸбҹ’бһҡбһ№бһҳбһҸбҹӮбһ“бҹ…бһӣбһҫбһ”бһүбҹ’бһҮбһ¶бһҮбһ¶бһҖбҹҶбһҺбһҸбҹӢ бһ¬бһ—бһ¶бһҹбһ¶бһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ‘бҹҒбҹ” бһ•бҹ’бһ‘бһ»бһҷбһ‘бҹ…бһңбһ·бһү бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһҸбһ¶бһҳбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҠбһҫбһҳбҹ’бһ”бһёбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбҹ” бһўбһҸбҹ’бһҗбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһӮбһә бһҠбҹ„бһҷбһҹбһ¶бһҡбһҸбҹӮбһўбҹ’бһ“бһҖбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӣбһҫбһҜбһҖбһҹбһ¶бһҡбһ…бҹ’бһҡбһҫбһ“ бһӮбһәбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹӮбһ”бһ”бһ»бһҡбһ¶бһҺбһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһӣбһҫбһўбҹ’бһңбһёбһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

### бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһўбҹ’бһңбһёбһ”бһ¶бһ“бһҒбҹ’бһӣбҹҮбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһҡбһҝбһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

- **бһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ–бһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶ chatbot**бҹ” бһҮбһ¶бһҡбһјбһ”бһ—бһ¶бһ–бһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһўбҹҶбһ–бһёбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘ бһҠбһјбһ…бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһңбһ¶ бһўбһ¶бһ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһӣбҹ’бһўбҹ”
- **бһҮбһҪбһҷбһўбҹ’бһ“бһҖ**бҹ” LLM бһҳбһ¶бһ“бһӣбһҖбҹ’бһҒбһҺбҹҲбһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһҠбһјбһ…бһҮбһ¶ бһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘ бһ‘бһ‘бһҪбһӣбһҷбһҖбһ…бҹҶбһҺбҹҒбҹҮбһҠбһ№бһ„бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶бһҡбҹҒбһҹбҹҠбһ»бһҳ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”
- **бһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠ**бҹ” бһ“бҹ…бһӣбһҫбһҡбһ”бҹҖбһ”бһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһҮбһ¶бһҳбһҪбһҷ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһңбһ¶бһӣбҹӮбһ„бһ”бһ¶бһ“бһ бһҫбһҷ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҡбҹ’бһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷ LLM бһҠбҹӮбһӣбһ—бһ¶бһӮбһ…бҹ’бһҡбһҫбһ“бһҳбһ¶бһ“бһҡбһ”бҹҖбһ”бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһ”бһҺбҹ’бһҠбһ¶бһүбһҮбһ¶бһҳбһҪбһҷ prompt бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ бһҫбһҷбһ‘бһ‘бһҪбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһ”бһ·бһ‘бһҖбһүбҹ’бһ…бһ”бҹӢбһҖбһ¶бһҡбһ бҹ… API бһ“бһ·бһ„бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDKs

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҹбҹ’бһӮбһ¶бһӣбҹӢбһӣбҹ’бһўбһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ LLMs бһҠбһјбһ…бһҮбһ¶бҹ–

- **openai** бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһҳбһ¶бһ“бһ—бһ¶бһ–бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ бһҫбһҷбһ•бҹ’бһүбһҫ prompt бһ…бһјбһӣбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһӣбһҫбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһ”бһ“бҹ’бһҸбһ·бһ…бһҠбһјбһ…бһҮбһ¶бҹ–

- **Langchain**бҹ” Langchain бһ•бҹ’бһҹбһ¶бһҡбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҖбҹ’бһ“бһ»бһ„ Pythonбҹ”
- **Semantic Kernel**бҹ” Semantic Kernel бһҮбһ¶бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ–бһё Microsoft бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ C#, Python, бһ“бһ·бһ„ Javaбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһ”бҹ’бһҡбһҫ openai

бһҳбһҖбһҳбһҫбһӣбһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһҷбһҫбһ„бһўбһ¶бһ…бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„ бһҡбһ”бҹҖбһ”бһҺбһ¶бһҳбһҪбһҷбһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ бһ“бһ·бһ„бһҸбһҳбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһ”бҹүбһ»бһ“бҹ’бһҳбһ¶бһ“бҹ”

### бһҸбҹҶбһЎбһҫбһ„ openai

  > [!NOTE] бһҮбҹҶбһ бһ¶бһ“бһ“бҹҒбҹҮбһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ‘бҹҶбһ–бҹҗбһҡбһ“бҹҒбҹҮбһӣбһҫ Codespaces бһ¬ Devcontainer


бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷ OpenAI бһ¬ Azure OpenAIбҹ” бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ C#, Python, JavaScript, Java бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”  
бһҷбһҫбһ„бһ”бһ¶бһ“бһҮбҹ’бһҡбһҫбһҹбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ Python `openai` бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `pip` бһҠбһҫбһҳбҹ’бһ”бһёбһҸбҹҶбһЎбһҫбһ„бһңбһ¶бҹ”

```bash
pip install openai
```

бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһ‘бҹҶбһ–бҹҗбһҡбһ“бҹҒбҹҮбһӣбһҫ Codespaces бһ¬ Dev Container бһ“бҹ„бҹҮбһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҸбҹҶбһЎбһҫбһ„ [Python](https://www.python.org/) бһӣбһҫбһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ•бһ„бҹ”

### бһ”бһ„бҹ’бһҖбһҫбһҸбһ’бһ“бһ’бһ¶бһ“ бһ“бһ·бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһ·бһ“бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһҡбһҪбһ… бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҮбҹҶбһ бһ¶бһ“бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бһ„бҹ’бһҖбһҫбһҸбһӮбһҺбһ“бһёбһ“бҹ…бһӣбһҫ OpenAI <https://platform.openai.com/signup>бҹ”
- бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһ‘бһ‘бһҪбһӣбһҷбһҖбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ <https://platform.openai.com/api-keys>бҹ”

1. бһ”бҹ’бһҡбһҫбһҖбһјбһҠ *app.py* бһҠбҹӮбһӣбһҳбһ¶бһ“бһҡбһҪбһ…бһҮбһ¶бһ–бҹ’бһҡбҹҶбһҠбҹӮбһ“бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ  
1. бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһўбһҗбҹҒбһҡ `prompt` бһ бһҫбһҷбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбһҖбһјбһҠбһҡбһ”бһҹбҹӢбһңбһ¶бһ‘бҹ…бһҮбһ¶бҹҲ


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
