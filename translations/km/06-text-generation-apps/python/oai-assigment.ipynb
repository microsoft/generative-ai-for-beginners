# бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһ‘бһӣбҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбһҸбһ¶бһҳбһҡбһҷбҹҲбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮбһҗбһ¶ бһҳбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ prompt бһ бһҫбһҷбһҳбһ¶бһ“бһңбһёбһ‘бҹ’бһҷбһ¶бһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡбһҳбһҪбһҷбһҲбҹ’бһҳбҹ„бҹҮбһҗбһ¶ "prompt engineering" бһҖбҹҸбһҳбһ¶бһ“бһҠбҹӮбһҡбҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһ”бһ¶бһ“ бһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ prompt бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһ”бһ¶бһ“бһўбҹ’бһңбһёбһҳбһҪбһҷбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһҠбһјбһ…бһ“бҹҒбҹҮбһ‘бҹ…бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһҠбһ№бһ„бһўбҹҶбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, completions бһ бһҫбһҷбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһ”бһҫбһҖбһ•бҹ’бһҠбҹҒбһҖ

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„:

- бһҡбҹҖбһ“бһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҷбһӣбҹӢбһҠбһ№бһ„бһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, temperature, бһ“бһ·бһ„ tokens бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҖбһ¶бһҡбһҹбһ·бһҖбҹ’бһҹбһ¶

бһ“бҹ…бһ…бһ”бҹӢбһңбһӮбҹ’бһӮбһҹбһ·бһҖбҹ’бһҹбһ¶бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…:

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһІбҹ’бһҷбһ”бҹ’бһҡбһҫ tokens бһ…бҹ’бһҡбһҫбһ“ бһ¬бһҸбһ·бһ… бһ бһҫбһҷбһҖбҹҸбһ”бҹ’бһҸбһјбһҡбһҸбһҳбҹ’бһӣбҹғ temperature бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһӣбһ‘бҹ’бһ’бһ•бһӣбһ“бһ¶бһ“бһ¶бҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһҮбһ¶бһўбҹ’бһңбһё?

бһҮбһ¶бһ‘бһјбһ‘бҹ… бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһңбһ¶бһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһҳбһҪбһҷбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ:

- бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ‘бһјбһ‘бҹ…бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһ“бһ·бһ„бһңбһ¶бһҹбһҳбҹ’бһҡбһ”бһҹбһҳбҹ’бһҡбһҪбһӣбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶бҹ”
- бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷбһ…бҹҶбһ“бһҪбһ“бһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһ… бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„, бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘, бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё Console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбһ¶бһҡбһҡбһ№бһҸбһ”бһ“бҹ’бһҸбһ№бһ„

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶ бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶:

- **бһңбһ¶бһҳбһ¶бһ“бһҖбһ¶бһҡбһҡбһ№бһҸбһ”бһ“бҹ’бһҸбһ№бһ„**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бһҺбһ¶бһҳбһҪбһҷбһ”бһ¶бһ“бһ‘бһ¶бҹҶбһ„бһўбһҹбҹӢбһ‘бҹҒ бһӮбһәбһҸбҹӮбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”
- **бһ—бһ¶бһҹбһ¶а№ҖаёүаёһаёІаё°**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһ‘бҹ„бҹҮбһ”бһёбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ¶бһ“бһҖбҹҸбһҠбҹ„бһҷбҹ”

### бһӮбһ»бһҺбһҹбһҳбҹ’бһ”бһҸбҹ’бһҸбһ·бһҡбһ”бһҹбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶бһ‘бҹ…бһҠбһјбһ…бһҳбҹ’бһҠбҹҒбһ…?

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбҹҒбһҡбһёбһ—бһ¶бһ–бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„ бһҳбһ·бһ“бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһҠбҹ„бһҷбһ”бһүбҹ’бһҮбһ¶бһҸбҹ’бһҡбһҪбһҸбһ–бһ·бһ“бһ·бһҸбҹ’бһҷбһ¬бһ—бһ¶бһҹбһ¶бһ…бһјбһӣбһ…бһ·бһҸбҹ’бһҸбһ‘бҹҒбҹ” бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһўбһ“бһ»бһүбҹ’бһүбһ¶бһҸбһІбҹ’бһҷбһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбҹ” бһӮбһ»бһҺбһҹбһҳбҹ’бһ”бһҸбҹ’бһҸбһ·бһҳбһҪбһҷбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһӮбһәбһҠбҹ„бһҷбһҹбһ¶бһҡбһҸбҹӮбһўбҹ’бһ“бһҖбһ бһҫбһҷбһҖбҹҶбһ–бһ»бһ„бһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӣбһҫбһҖбһ¶бһҡбһ”бҹ’бһҡбһҳбһјбһӣбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһҷбҹүбһ¶бһ„бһ…бҹ’бһҡбһҫбһ“ бһҒбһҺбҹҲбһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹғбһҺбһёбһўбһ¶бһ…бһҳбһ¶бһ“бһҖбһ¶бһҡбһҡбһ№бһҸбһ”бһ“бҹ’бһҸбһ№бһ„бһӣбһҫбһўбҹ’бһңбһёбһҠбҹӮбһӣбһҳбһ¶бһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

### бһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһўбҹ’бһңбһёбһ”бһ¶бһ“бһҒбҹ’бһӣбҹҮбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ:

- **бһҮбһ¶бһҡбһјбһ”бһҳбһ“бҹ’бһҸвҖӢбһ”бҹ’бһҡбһ¶бһҹбҹ’бһҡбҹҗбһҷбһҸбһ¶бһҳбһ”бһ…бҹ’бһ…бҹҒбһҖбһңбһ·бһ‘бҹ’бһҷбһ¶**бҹ” бһҮбһ¶бһҡбһјбһ”бһҳбһ“бҹ’бһҸбһ”бҹ’бһҡбһ¶бһҹбҹ’бһҡбҹҗбһҷбһҸбһ”бһҹбҹҶбһҺбһҪбһҡбһ–бһ¶бһҖбҹӢбһ–бҹҗбһ“бҹ’бһ’бһ“бһ№бһ„бһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘ бһҠбһјбһ…бһҮбһ¶бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһўбһ¶бһ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһӣбҹ’бһўбҹ”
- **бһҮбһҪбһҷбһҖбһ¶бһҡ**бҹ” LLMs бһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҡбһҝбһ„бһҠбһјбһ…бһҮбһ¶ бһҜбһҖбһҹбһ¶бһҡбһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘, бһ‘бһ‘бһҪбһӣбһҷбһҖбһ…бҹҶбһҺбҹҒбҹҮбһҠбһ№бһ„бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘, бһ•бһӣбһ·бһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶ CV бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”
- **бһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠ**бҹ” бһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһ“бһ·бһ„ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһ”бһ¶бһ“бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫ бһ…бһ„бҹӢбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҮбһ¶бһҳбһҪбһҷ LLM бһ—бһ¶бһӮбһ…бҹ’бһҡбһҫбһ“бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡбһңбһ·бһ’бһёбһҡбҹҖбһ”бһ…бҹҶбһ–бһёбһҡбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҠбһјбһ…бһҮбһ¶:

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһ”бһҺбҹ’бһҸбһ¶бһүбһҠбҹӮбһӣбһҳбһ¶бһ“ prompt бһ бһҫбһҷбһ‘бһ‘бһҪбһӣбһўбҹ’бһ“бһҖбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһЎбһҫбһ„бһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһҹбҹ’бһҡбһҪбһӣбһҖбһ¶бһҡбһ бҹ… API бһ“бһ·бһ„бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDKs

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһҹбҹ’бһӮбһ¶бһӣбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ LLMs бһҠбһјбһ…бһҮбһ¶:

- **openai**, бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһҹбҹ’бһҡбһҪбһӣбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бҹ’бһүбһҫ promptбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹҒбҹҮбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһӣбһҫбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҮбһ¶бһ„ бһҠбһјбһ…бһҮбһ¶:

- **Langchain**. Langchain бһ”бһ¶бһ“бһҹбҹ’бһӮбһ¶бһӣбҹӢбһӣбҹ’бһў бһ“бһ·бһ„бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ Pythonбҹ”
- **Semantic Kernel**. Semantic Kernel бһҮбһ¶бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһ–бһё Microsoft бһҠбҹӮбһӣбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ C#, Python, бһ“бһ·бһ„ Javaбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһ”бҹ’бһҡбһҫ openai

бһҳбһҖбһҳбһҫбһӣбһҗбһ¶бһҸбһҫбһҷбһҫбһ„бһўбһ¶бһ…бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ… бһҸбһҫбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһўбҹ’бһңбһё бһўбҹ’бһңбһёбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ¶бһ“бҹӢ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”

### бһҸбһҳбҹ’бһӣбһҫбһ„ openai

  > [!NOTE] бһҮбҹҶбһ бһ¶бһ“бһ“бҹҒбҹҮбһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҡбһҸбҹӢ notebook бһ“бҹҒбҹҮбһӣбһҫ Codespaces бһ¬бһҖбҹ’бһ“бһ»бһ„ Devcontainer бһ‘бҹҒ


бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһҹбһ бһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ OpenAI бһ¬ Azure OpenAIбҹ” бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ C#, Python, JavaScript, Java бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”  
бһҷбһҫбһ„бһ”бһ¶бһ“бһҮбҹ’бһҡбһҫбһҹбһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ Python `openai` бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `pip` бһҠбһҫбһҳбҹ’бһ”бһёбһҸбһҳбҹ’бһӣбһҫбһ„бһңбһ¶бҹ”

```bash
pip install openai
```

### бһ”бһ„бҹ’бһҖбһҫбһҸбһ’бһ“бһ’бһ¶бһ“ бһ“бһ·бһ„бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһ·бһ“бһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһ бһҫбһҷ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһўбһ“бһ»бһңбһҸбҹ’бһҸбһҮбҹҶбһ бһ¶бһ“бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳ:

- бһ”бһ„бҹ’бһҖбһҫбһҸбһӮбһҺбһ“бһёбһ“бҹ… OpenAI <https://platform.openai.com/signup>бҹ”
- бһҘбһЎбһјбһңбһҹбһјбһҳбһ‘бһ‘бһҪбһӣбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ <https://platform.openai.com/api-keys>бҹ” 

>[!NOTE]
> бһңбһ¶бһӮбһҪбһҡбһІбҹ’бһҷбһ…бҹӮбһҖбһ…бһ¶бһҷбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ–бһёбһҖбһјбһҠбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ environment variablesбҹ”
> - бһҖбҹҶбһҺбһҸбҹӢ environment variable `OPENAI_KEY` бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ .envбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһӣбҹҶбһ бһ¶бһҸбҹӢбһҖбһ¶бһӣбһ–бһёбһҳбһ»бһ“бһ“бҹғбһңбһӮбҹ’бһӮбһ“бҹҒбҹҮбһ бһҫбһҷ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҖбһ¶бһҡбһҡбҹҖбһ”бһ…бҹҶбһҡбһҪбһ…бһҡбһ¶бһӣбҹӢбҹ”
> - бһңбһ¶бһҳбһ¶бһ“бһҹбһ¶бһҡбһҹбҹҶбһҒбһ¶бһ“бҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҖбһҸбҹӢбһҹбһҳбҹ’бһ–бһ¶бһ’бһҗбһ¶ бһҖбһјбһ“бһҹбҹ„ API бһ“бһ№бһ„бһўбһ¶бһ…бһ…бһјбһӣбһ”бҹ’бһҡбһҫбһ”бһ¶бһ“бһҸбҹ’бһҡбһ№бһҳбһҸбҹӮбһҳбҹ’бһҠбһ„бһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ” бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҹбһјбһҳбһ”бҹ’бһҡбһ¶бһҖбһҠбһҗбһ¶бһңбһ¶бһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ…бһҳбҹ’бһӣбһ„бһҸбҹ’бһҡбһ№бһҳбһҸбҹ’бһҡбһјбһңбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһңбһ¶бһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһҠбһјбһ…бһҠбҹӮбһӣбһҡбһ¶бһ”бҹӢбһўбһ¶бһ“бһ‘бҹҒ бһӮбһҪбһҡбһӣбһ»бһ”бһҖбһјбһ“бһҹбҹ„бһ“бһ·бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҗбҹ’бһҳбһёбҹ”


### бһҡбҹҖбһ”бһ…бҹҶбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’ OpenAI

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ OpenAI бһ“бҹҒбҹҮбһҮбһ¶бһңбһ·бһ’бһёбһҡбҹҖбһ”бһ…бҹҶбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

бһҒбһ¶бһ„бһӣбһҫ бһҷбһҫбһ„бһҖбҹҶбһ–бһ»бһ„бһҖбҹҶбһҺбһҸбҹӢбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- `api_key`, бһ“бҹҒбҹҮбһҮбһ¶бһҖбһјбһ“бһҹбҹ„ API бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҠбҹӮбһӣбһ”бһ¶бһ“бһғбһҫбһүбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһ•бҹ’бһ‘бһ¶бҹҶбһ„ OpenAIбҹ”
- `deployment`, бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһҮбҹҶбһ“бһ¶бһ“бҹӢ GPT бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ”

> [!NOTE]
> `os.environ` бһӮбһәбһҮбһ¶бһўбһ“бһ»бһӮбһҳбһ“бҹҚбһҠбҹӮбһӣбһўбһ¶бһ“ environment variablesбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһңбһ¶бһҠбһҫбһҳбҹ’бһ”бһёбһўбһ¶бһ“ environment variables бһҠбһјбһ…бһҮбһ¶ `OPENAI_API_KEY`бҹ”

## бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһ”бҹ’бһҡбһҫбһңбһ·бһ’бһёбһҹбһ¶бһҹбҹ’бһҸбҹ’бһҡ `responses.create`бҹ” бһ“бҹҒбҹҮбһҮбһ¶бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбһҒбһ¶бһ„бһӣбһҫ бһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһҸбҹ’бһҗбһ»бһ…бҹҶбһӣбһҫбһҷ бһ бһҫбһҷбһ•бҹ’бһүбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһ”бҹ’бһҡбһҫ бһ“бһ·бһ„ prompt бҹ” бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ”бҹ„бҹҮбһ–бһ»бһҳбҹ’бһ–бһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбҹ”

### бһ…бҹҶбһӣбһҫбһҷбһҮбһҮбҹӮбһҖ

бһҡбһ бһјбһҸбһҳбһҖбһҠбһӣбҹӢбһ”бһ…бҹ’бһ…бһ»бһ”бҹ’бһ”бһ“бҹ’бһ“ бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ–бһё prompt бһҸбҹӮбһҳбһҪбһҷбҹ” Responses API бһҖбҹҸбһӣбҹ’бһўбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢ chatbots бһ•бһ„бһҠбҹӮбһҡ - бһўбҹ’бһ“бһҖбһ•бҹ’бһҸбһӣбҹӢбһ”бһүбҹ’бһҮбһёбһҹбһ¶бһҡвҖӢбһҮбһ¶ `input` бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһҹбһ“бҹ’бһ‘бһ“бһ¶бҹ” бһ“бҹҒбҹҮбһҮбһ¶бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбһ“бҹғбһҖбһ¶бһҡбһ”бҹ’бһҡбһҫбһңбһ¶бҹ–

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## бһӣбҹҶбһ бһ¶бһҸбҹӢ - бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ

бһҘбһЎбһјбһңбһ“бҹҒбҹҮбһҷбһҫбһ„бһ”бһ¶бһ“бһҡбҹҖбһ“бһ–бһёбһҡбһ”бҹҖбһ”бһҸбҹҶбһЎбһҫбһ„ бһ“бһ·бһ„бһҡбҹҖбһ”бһ…бҹҶбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳ OpenAI бһҡбһҪбһ…бһ бһҫбһҷ бһ–бҹҒбһӣбһ“бҹҒбҹҮбһӮбһәбһ–бҹҒбһӣбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹҶбһ”бһјбһ„бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҸбһ¶бһҳбһҮбҹҶбһ бһ¶бһ“бһ‘бһ¶бҹҶбһ„бһ“бҹҒбҹҮбҹ–

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

1. бһ”бҹ’бһҡбһҫбһҜбһҖбһҹбһ¶бһҡ *app.py* бһҠбҹӮбһӣбһҳбһ¶бһ“бһҡбһҪбһ…бһҮбһ¶бһ…бҹҶбһҺбһ»бһ…бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳ
1. бһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһўбһҗбҹҒбһҡ `prompt` бһ бһҫбһҷбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҠбһјбһҡбһҖбһјбһҠбһҡбһ”бһҹбҹӢбһңбһ¶бһ‘бҹ…бһҮбһ¶бҹҲ


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
