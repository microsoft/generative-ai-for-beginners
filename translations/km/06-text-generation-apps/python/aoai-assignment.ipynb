# សាងសង់កម្មវិធីបង្កើតអត្ថបទ

អ្នកបានឃើញរហូតមកដល់គន្លងកម្មវិធីនេះថាមានគំនិតមូលដ្ឋានដូចជាពាក្យបញ្ចូល និងមានវិស័យមួយទៀតហៅថា "វិស្វកម្មពាក្យបញ្ចូល"។ ឧបករណ៍ជាច្រើនដែលអ្នកអាចធ្វើអន្តរប្រតិកម្មបានដូចជា ChatGPT, Office 365, Microsoft Power Platform និងផ្សេងទៀត គាំទ្រអ្នកដោយការប្រើពាក្យបញ្ចូលដើម្បីបំពេញភារកិច្ចមួយ។

សម្រាប់អ្នកដើម្បីបន្ថែមបទពិសោធន៍ដូចនេះទៅកម្មវិធីមួយ អ្នកត្រូវយល់ដឹងពីគំនិតដូចជាពាក្យបញ្ចូល, ការសម្រេចចិត្ត និងជ្រើសរើសបណ្ណាល័យមួយដើម្បីធ្វើការជាមួយ។ នេះជាការចាំបាច់ដែលអ្នកនឹងរៀននៅក្នុងជំពូកនេះ។

## ការណែនាំ

នៅក្នុងជំពូកនេះ អ្នកនឹង៖

- រៀនអំពីបណ្ណាល័យ openai និងគំនិតមូលដ្ឋានរបស់វា។
- សាងសង់កម្មវិធីបង្កើតអត្ថបទដោយប្រើ openai។
- យល់ដឹងពីរបៀបប្រើគំនិតដូចជា ពាក្យបញ្ចូល, សីតុណ្ហភាព និងតូនកិន ដើម្បីសាងសង់កម្មវិធីបង្កើតអត្ថបទ។

## គោលបំណងរៀន

នៅចុងមេរៀននេះ អ្នកនឹងអាច៖

- ពន្យល់ថាកម្មវិធីបង្កើតអត្ថបទគឺជាអ្វី។
- សាងសង់កម្មវិធីបង្កើតអត្ថបទដោយប្រើ openai។
- កំណត់កម្មវិធីរបស់អ្នកឲ្យប្រើតូនកិនច្រើន ឬតិច និងផ្លាស់ប្ដូរសីតុណ្ហភាព សម្រាប់លទ្ធផលដែលមានការប្រែប្រួល។

## កម្មវិធីបង្កើតអត្ថបទគឺជាអ្វី?

ធម្មតា ក្រោយពីអ្នកសាងសង់កម្មវិធីហើយ វាមានចំណុចបញ្ចូលមួយប្រភេទដូចខាងក្រោម៖

- កម្មវិធីមូលដ្ឋានបញ្ជា។ កម្មវិធី Console ជាកម្មវិធីនិយមមួយដែលអ្នកវាយបញ្ជារ ហើយវា​អនុវត្តភារកិច្ចមួយ។ ឧទាហរណ៍ `git` គឺជាកម្មវិធីផ្អែកលើបញ្ជា។
- ចំណុចបញ្ចូលអ្នកប្រើ (UI)។ កម្មវិធីខ្លះមានចំណុចបញ្ចូលក្រាហ្វិក ដែលអ្នកចុចប៊ូតុង វាយអត្ថបទ ជ្រើសជម្រើស និងផ្សេងទៀត។

### កម្មវិធី Console និង UI មានកំរិតកំណត់

ប្រៀបធៀបវាជាមួយកម្មវិធីផ្អែកលើបញ្ជា ដែលអ្នកវាយបញ្ជារ៖

- **វាមានកំរិតកំណត់**។ អ្នកមិនអាចវាយបញ្ជារណាមួយបានទេ តែគឺតែបញ្ជារដែលកម្មវិធីគាំទ្រតែប៉ុណ្ណោះ។
- **ភាសារពិសេស**។ កម្មវិធីខ្លះគាំទ្រភាសាច្រើន ប៉ុន្តែកម្មវិធីប្រើប្រាស់លំនាំដើមសម្រាប់ភាសាពិសេសមួយ ទោះបីលើកអាចបន្ថែមគាំទ្រភាសាបន្ថែមផង។

### ឧបករណ៍នៃកម្មវិធីបង្កើតអត្ថបទ

តើយ៉ាងដូចម្តេចដែលកម្មវិធីបង្កើតអត្ថបទមានភាពខុសគ្នា?

នៅក្នុងកម្មវិធីបង្កើតអត្ថបទ អ្នកមានភាពបត់បែនច្រើន ការពិតណាស់ អ្នកមិនបានកំណត់នៅក្នុងបញ្ជារណាមួយ ឬភាសាប្រភេទណាមួយទេ។ ទឹកចិត្តវិញ អ្នកអាចប្រើភាសាធម្មជាតិសម្រាប់ធ្វើអន្តរប្រតិកម្មជាមួយកម្មវិធីបាន។ អត្ថប្រយោជន៍ផ្សេងទៀតគឺដោយសារអ្នកកំពុងធ្វើអន្តរប្រតិកម្មជាមួយប្រភពទិន្នន័យដែលបានហ្វឹកហាត់លើចំណេះដឹងធំ មួយ ក្រៅពីកម្មវិធីទំនើបអាចមានកំណត់លើអ្វីដែលមាននៅក្នុងមូលដ្ឋានទិន្នន័យ។

### តើខ្ញុំអាចសង់អ្វីបានជាមួយកម្មវិធីបង្កើតអត្ថបទ?

មានរឿងជាច្រើនដែលអ្នកអាចសាងសង់បាន។ ឧទាហរណ៍៖

- **កម្មវិធីជជែកបណ្ដាញបែប chatbot**។ កម្មវិធីបញ្ជូនសំណួរអំពីប្រធានបទ ដូចជាក្រុមហ៊ុន និងផលិតផលរបស់អ្នកអាចជាចម្លើយល្អ។
- **ជំនួយករ**។ LLMs ល្អសម្រាប់ការសង្រ្គោះអត្ថបទ រកយកចំណុចដែលសំខាន់ ព្យាយាមផលិតអត្ថបទដូចជា ប្រវត្តិរូប និងផ្សេងទៀត។
- **ជំនួយកូដ**។ អាស្រ័យលើម៉ូឌែលភាសាដែលអ្នកប្រើ អ្នកអាចសាងសង់ជំនួយកូដដែលជួយអ្នកសរសេរកូដ។ ឧទាហរណ៍ អ្នកអាចប្រើផលិតផលដូចជា GitHub Copilot និង ChatGPT ដើម្បីជួយសរសេរកូដ។

## តើខ្ញុំអាចចាប់ផ្តើមយ៉ាងដូចម្តេច?

យ៉ាងហោចណាស់ អ្នកត្រូវស្វែងរកវិធីសាស្រ្តសម្រាប់ផ្សំភ្ជាប់ជាមួយ LLM ដែលភាគច្រើនរួមមានរបៀបពីរខាងក្រោម៖

- ប្រើ API។ នៅទីនេះ អ្នកកំពុងបង្កើតការស្នើសុំបណ្ដាញជាមួយពាក្យបញ្ចូលរបស់អ្នក ហើយទទួលអត្ថបទដែលបានបង្កើតត្រឡប់មកវិញ។
- ប្រើបណ្ណាល័យ។ បណ្ណាល័យជួយបង្ហាប់ការហៅ API ហើយធ្វើឲ្យងាយស្រួលប្រើ។

## បណ្ណាល័យ/SDKs

មានបណ្ណាល័យមួយចំនួនល្បីសម្រាប់ធ្វើការជាមួយ LLM ដូចជា៖

- **openai**, បណ្ណាល័យនេះធ្វើឲ្យងាយស្រួលក្នុងការតភ្ជាប់ទៅម៉ូឌែលរបស់អ្នក និងផ្ញើពាក្យបញ្ចូល។

បន្ទាប់មកមានបណ្ណាល័យដែលដំណើរការលើកម្រិតខ្ពស់ជាច្រើនដូចជា៖

- **Langchain**។ Langchain មានឈ្មោះល្បីនិងគាំទ្រ Python។
- **Semantic Kernel**។ Semantic Kernel ជាបណ្ណាល័យពី Microsoft គាំទ្រភាសា C#, Python, និង Java។

## កម្មវិធីដំបូងប្រើ openai

មកមើលរបៀបដែលយើងអាចសាងសង់កម្មវិធីដំបូងរបស់យើង តើបណ្ណាល័យណាប្រើ តើត្រូវការ​ប៉ុន្មាន និងផ្សេងទៀត។

### តម្លឹង openai

  > [!NOTE] ជំហាននេះមិនចាំបាច់បើបើកចំហតែបណ្ណាល័យនេះនៅ Codespaces ឬក្នុង Devcontainer


មានបណ្ណាល័យជាច្រើនសម្រាប់ធ្វើអន្តរប្រតិកម្មជាមួយ OpenAI ឬ Azure OpenAI។ អាចប្រើភាសាកម្មវិធីជាច្រើនដូចជា C#, Python, JavaScript, Java និងផ្សេងទៀត។  
យើងបានជ្រើសរើសប្រើបណ្ណាល័យ Python ឈ្មោះ `openai` ដូច្នេះយើងនឹងប្រើ `pip` ដើម្បីតម្លើងវា។

```bash
pip install openai
```

បើអ្នកមិនបានដំណើរការកំណត់ត្រានេះនៅលើ Codespaces ឬ Dev Container អ្នកត្រូវតែដំឡើង [Python](https://www.python.org/) លើម៉ាស៊ីនរបស់អ្នកផងដែរ។

### បង្កើតធនធាន

ប្រសិនបើអ្នកមិនធ្វើរួចទេ អ្នកត្រូវអនុវត្តជំហានដូចខាងក្រោម៖

- បង្កើតគណនីនៅ Azure <https://azure.microsoft.com/free/>។
- ទទួលបានការចូលប្រើ Azure OpenAI។ ចូលទៅ <https://learn.microsoft.com/azure/ai-services/openai/overview#how-do-i-get-access-to-azure-openai> ហើយស្នើសុំការចូលប្រើ។

  > [!NOTE]
  > បច្ចុប្បន្នពេលនោះ អ្នកត្រូវដាក់ពាក្យសុំចូលប្រើ Azure OpenAI។

- បង្កើតធនធាន Azure OpenAI Service។ អានមគ្គុទេសក៍នេះសម្រាប់របៀប [បង្កើតធនធាន](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst)។

### រកកូនសោ API និងចំណុចបញ្ចប់

នៅពេលនេះ អ្នកត្រូវប្រាប់បណ្ណាល័យ `openai` របស់អ្នកអំពីកូនសោ API ដែលត្រូវប្រើ។ ដើម្បីស្វែងរកកូនសោ API របស់អ្នក ចូលទៅផ្នែក "Keys and Endpoint" នៃធនធាន Azure OpenAI របស់អ្នក ហើយចម្លងតម្លៃ "Key 1"។

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ឥឡូវនេះអ្នកមានព័ត៌មាននេះចម្លងរួចហើយ តោះបង្ហាញបណ្ណាល័យឲ្យប្រើវា។

1. ស្វែងរកអថេរ `prompt` ហើយផ្លាស់ប្តូរកូដរបស់វាទៅជាដូចខាងក្រោម៖


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

ល្អណាស់ យើងមកមើលថាតើយើងអាចធ្វើឲ្យរឿងៗកាន់តែល្អប្រសើរយ៉ាងដូចម្តេច។ ដើម្បីធ្វើឲ្យរឿងៗកាន់តែល្អប្រសើរ យើងចង់ធានាថាកូដមានភាពបត់បែន ដូច្នេះគ្រឿងផ្សំ និងចំនួនរូបមន្តអាចត្រូវបានធ្វើឲ្យប្រសើរ និងផ្លាស់ប្តូរ។


1. តោះផ្លាស់ប្តូរកូដទៅតាមរបៀបដូចខាងក្រោម:


````python
import os
from openai import AzureOpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = AzureOpenAI(
  azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"], 
  api_key=os.environ['AZURE_OPENAI_API_KEY'],  
  api_version = "2023-10-01-preview"
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
