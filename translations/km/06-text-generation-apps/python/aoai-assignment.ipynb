# бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһ‘бһӣбҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбһҸбһ¶бһҳбһңбһ·бһүбҹ’бһүбһ¶бһҹбһ¶бһ“бҹҒбҹҮбһҗбһ¶бһҳбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ prompts бһ бһҫбһҷбһҳбһ¶бһ“бһҖбҹҸбһҹбһ¶бһ’бһ¶бһҡбһҺбҹҲбһҳбһҪбһҷбһ‘бҹҖбһҸбһ бҹ…бһҗбһ¶ "prompt engineering"бҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһўбҹ’бһ“бһҖбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бҹ’бһҡбһҫ prompts бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһҳбҹ’бһҡбҹҒбһ…бһўбҹ’бһңбһёбһҳбһҪбһҷбҹ”

бһҠбһҫбһҳбҹ’бһ”бһёбһІбҹ’бһҷбһўбҹ’бһ“бһҖбһ”бһүбҹ’бһ…бһјбһӣбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһ”бҹӮбһ”бһ“бҹҒбҹҮбһ‘бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҳбһҪбһҷ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһ–бһёбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompts, completions бһ бһҫбһҷбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбҹ” бһ“бҹҒбҹҮбһҮбһ¶бһҖбһ¶бһҡбһ–бһ·бһҸбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”

## бһҳбһ»бһҒбһҳбһ¶бһҸбҹӢ

бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бҹ–

- бһҡбҹҖбһ“бһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ openaiбҹ”
- бһҷбһӣбҹӢбһҗбһ¶бһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһҫбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, temperature бһ“бһ·бһ„ tokens бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһҠбҹ…бһҖбһ¶бһҡбһҡбҹҖбһ“

бһ“бҹ…бһ…бһ»бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһ“бҹғбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһўбҹ’бһңбһёбһ‘бҹ…бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”
- бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһІбҹ’бһҷбһ”бҹ’бһҡбһҫ tokens бһ…бҹ’бһҡбһҫбһ“ бһ¬бһҸбһ·бһ… бһ“бһ·бһ„бһ”бҹ’бһҠбһјбһҡ temperature бһҠбһҫбһҳбҹ’бһ”бһёбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһӣбһ‘бҹ’бһ’бһ•бһӣбһ•бҹ’бһҹбҹҒбһ„бҹ—бһӮбҹ’бһ“бһ¶бҹ”

## бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҮбһ¶бһўбҹ’бһңбһё?

бһҮбһ¶бһ‘бһјбһ‘бҹ… бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһҳбһҪбһҷбһңбһ¶бһҳбһ¶бһ“бһҳбһ»бһҒбһ„бһ¶бһҡбһҡбһјбһ”бһ—бһ¶бһ–бһҳбһҪбһҷбһ”бҹӮбһ”бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһ–бҹҗбһ“бҹ’бһ’бһ”бһүбҹ’бһҮбһ¶ (Command-based)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё Console бһӮбһәбһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһ“бһ№бһ„бһўбһ“бһ»бһңбһҸбҹ’бһҸбһҖбһ¶бһҡбһ„бһ¶бһҡбһҳбһҪбһҷбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһҫбһ”бһүбҹ’бһҮбһ¶бҹ”
- бһҳбһ»бһҒбһ„бһ¶бһҡбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢ (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“ graphical user interfaces (GUIs) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһ”бһүбҹ’бһ…бһјбһӣбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё Console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһ‘бҹ…бһҖбһ¶бһ“бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһҫбһ”бһүбҹ’бһҮбһ¶ бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бҹ–

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бһҺбһ¶бһҳбһҪбһҷбһ”бһ¶бһ“бһ‘бҹҒ бһҳбһ¶бһ“бһҸбҹӮбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”
- **бһ—бһ¶бһҹбһ¶бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһҠбһјбһ…бһҮбһ¶бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ”бһ“бҹ’бһҗбҹӮбһҳбһ‘бҹҖбһҸбһ”бһ¶бһ“бҹ”

### бһ”бҹ’бһҡбһ¶бһҖбҹӢбһ…бҹҶбһҺбҹҒбһүбһ–бһёбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҸбһҫбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҳбһ¶бһ“бһ—бһ¶бһ–бһҒбһ»бһҹбһӮбҹ’бһ“бһ¶?

бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҜбһҖбһҹбһ·бһ‘бҹ’бһ’бһ·бһ…бҹ’бһҡбһҫбһ“ бһ бһҫбһҷбһҳбһ·бһ“бһҒбһ·бһҸбһҖбҹҶбһҺбһҸбҹӢбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҠбҹ„бһҷбһҖбҹҶбһҺбһҸбҹӢбһ”бһүбҹ’бһҮбһ¶ бһ¬бһ—бһ¶бһҹбһ¶ бһ“бҹ„бҹҮбһ‘бҹҒбҹ” бһ•бҹ’бһ‘бһ»бһҷбһ‘бҹ…бһңбһ·бһү бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ”бҹ’бһҡбһҖбһ”бһҠбҹ„бһҷбһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ¶бһ“бҹ” бһ•бһ„бһҠбҹӮбһҡ бһҖбҹҸбһҠбҹ„бһҷбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҺбҹ’бһҸбһ»бҹҮбһ”бһҺбҹ’бһҸбһ¶бһӣбһӣбһҫбһҳбһ¶бһҸбһ·бһҖбһ¶бһ’бҹҶбһ‘бһјбһӣбһ¶бһҷ бһҒбһҺ: бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹ’бһҡбһ–бҹғбһҺбһёбһҳбһҪбһҷбһ“бҹ…бһҸбҹӮбһӣбҹҶбһ”бһ¶бһҖбһ“бҹ…бһӣбһҫбһўбҹ’бһңбһёбһҠбҹӮбһӣбһҳбһ¶бһ“бһҖбҹ’бһ“бһ»бһ„бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”

### бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһҹбһ„бҹӢбһўбҹ’бһңбһёбһ”бһ¶бһ“бһҒбҹ’бһӣбҹҮбһ–бһёбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһўбҹ’бһңбһёбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ„бҹӢбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

- **бһҮбһ¶бһӮбһјбһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶ (chatbot)**бҹ” бһҮбһ¶бһӮбһјбһ–бһ·бһ—бһ¶бһҖбҹ’бһҹбһ¶бһҠбҹӮбһӣбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһўбҹҶбһ–бһёбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘ бһҠбһјбһ…бһҮбһ¶ бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһңбһ¶ бһ”бһ¶бһ“бһӣбҹ’бһўбҹ”
- **бһҮбһ¶бһўбҹ’бһ“бһҖбһҮбһҪбһҷ**бҹ” LLMs бһҳбһ¶бһ“бһ–бһ“бҹ’бһӣбһәбһӣбҹ’бһўбһ“бҹ…бһӣбһҫбһҡбһҝбһ„бһҠбһјбһ…бһҮбһ¶ бһҹбһ¶бһҡбһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘ бһҷбһҖбһ…бҹҶбһҺбҹҒбҹҮбһҠбһ№бһ„бһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ•бһӣбһ·бһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶ resume бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбҹ”
- **бһҮбһ¶бһўбҹ’бһ“бһҖбһҮбһҪбһҷбһўбһҸбҹ’бһҗбһ”бһ‘бһҖбһјбһҠ**бҹ” бһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ„бҹӢбһўбҹ’бһ“бһҖбһҮбһҪбһҷбһҖбһјбһҠбһ”бһ¶бһ“ бһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһ“бһ·бһ„ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҷбҹүбһ¶бһ„бһҠбһјбһ…бһҳбҹ’бһҸбҹҒбһ…?

бһўбҹ’бһ“бһҖбһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһҸбҹ’бһҡбһјбһңбһҹбҹ’бһңбҹӮбһ„бһҡбһҖбһңбһ·бһ’бһёбһ’бҹ’бһңбһҫбһҮбһ¶бһҳбһҪбһҷ LLM бһҠбҹӮбһӣбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһ“бҹ…бһ–бһёбһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҹбҹҶбһҺбһҫбһўбҹҠбһёбһ“бһ’бһәбһҺбһ·бһҸбһҮбһ¶бһҳбһҪбһҷ prompt бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ бһҫбһҷбһ‘бһ‘бһҪбһӣбһ”бһ¶бһ“бһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ„бҹ’бһҖбһҫбһҸбһЎбһҫбһ„бһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһ”бһ·бһ‘бһ”бһ¶бҹҶбһ„бһҖбһ¶бһҡбһ бҹ… API бһ бһҫбһҷбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDKs

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҹбҹ’бһӮбһ¶бһӣбҹӢбһӮбҹ’бһҡбһ”бҹӢбһӮбҹ’бһҡбһ¶бһ“бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ LLMs бһҠбһјбһ…бһҮбһ¶бҹ–

- **openai** бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһІбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бҹ’бһүбһҫ prompt бһ…бһјбһӣбҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһӣбһҫбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҮбһ¶бһ„бҹ–

- **Langchain**бҹ” Langchain бһҳбһ¶бһ“бһҹбһҖбҹ’бһҸбһ¶бһ“бһ»бһ–бһӣбһ“бһ·бһ„бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶ Pythonбҹ”
- **Semantic Kernel**бҹ” Semantic Kernel бһҮбһ¶бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹ„бһҷ Microsoft бһӮбһ¶бҹҶбһ‘бҹ’бһҡ C#, Python, бһ“бһ·бһ„ Javaбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһ”бҹ’бһҡбһҫ openai

бһҸбҹ„бҹҮбһҳбһҫбһӣбһ–бһёбһҡбһ”бҹҖбһ”бһҹбһ¶бһ„бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„ бһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһҷбһҫбһ„бһҸбҹ’бһҡбһјбһңбһҖбһ¶бһҡ бһ”бһҡбһ·бһҳбһ¶бһҺбһҠбҹӮбһӣбһ‘бһ¶бһҳбһ‘бһ¶бһҡ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”

### бһҠбҹҶбһЎбһҫбһ„ openai

  > [!NOTE] бһҮбҹҶбһ бһ¶бһ“бһ“бҹҒбҹҮбһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ бҹ… notebook бһ“бҹҒбҹҮбһ“бҹ… Codespaces бһ¬бһҖбҹ’бһ“бһ»бһ„ Devcontainer бһ‘бҹҒ


бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ OpenAI бһ¬ Azure OpenAIбҹ” бһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһҖбһјбһҠбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбһјбһ…бһҮбһ¶ C#, Python, JavaScript, Java бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸбһ”бһ¶бһ“бҹ”  
бһҷбһҫбһ„бһ”бһ¶бһ“бһҮбҹ’бһҡбһҫбһҹбһ…бҹҒбһү `openai` бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ Python бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮбһҷбһҫбһ„бһ“бһ№бһ„бһ”бҹ’бһҡбһҫ `pip` бһҠбһҫбһҳбҹ’бһ”бһёбһҠбҹҶбһЎбһҫбһ„бһңбһ¶бҹ”

```bash
pip install openai
```

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһҳбһ·бһ“бһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡ notebook бһ“бҹҒбҹҮбһ“бҹ… Codespaces бһ¬ Dev Container бһ‘бҹҒ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҠбҹҶбһЎбһҫбһ„ [Python](https://www.python.org/) бһ“бҹ…бһӣбһҫбһҳбҹүбһ¶бһҹбҹҠбһёбһ“бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ•бһ„бһҠбҹӮбһҡбҹ”

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
