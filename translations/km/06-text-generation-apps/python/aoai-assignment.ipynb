# бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһғбһҫбһүбһҡбһ бһјбһҸбһҳбһҖбһҠбһӣбҹӢбһ–бҹҒбһӣбһ“бҹҒбҹҮбһҸбһ¶бһҳбһҡбһҷбҹҲбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮбһҗбһ¶бһҳбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ prompt бһ“бһ·бһ„бһҹбһ¶бһҡбһ–бһҫбһ—бҹҗбһҺбҹ’бһҢбһҳбһҪбһҷбһ‘бһ¶бҹҶбһ„бһҳбһјбһӣбһҠбҹӮбһӣбһ бҹ…бһҗбһ¶ "prompt engineering"бҹ” бһ§бһ”бһҖбһҡбһҺбҹҚбһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷ бһҠбһјбһ…бһҮбһ¶ ChatGPT, Office 365, Microsoft Power Platform бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸ бһӮбһ¶бҹҶбһ‘бҹ’бһҡбһўбҹ’бһ“бһҖбһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ prompt бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһүбҹ’бһ…бһ”бҹӢбһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһҳбһҪбһҷбҹ”  

бһҠбһҫбһҳбҹ’бһ”бһёбһўбҹ’бһ“бһҖбһ”бһүбҹ’бһ…бһјбһӣбһ”бһ‘бһ–бһ·бһҹбҹ„бһ’бһ“бҹҚбһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“бһ“бҹҒбҹҮбһ‘бҹ…бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҷбһӣбҹӢбһ–бһёбһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһҠбһјбһ…бһҮбһ¶ prompt, completions бһ бһҫбһҷбһҮбҹ’бһҡбһҫбһҹбһҡбһҫбһҹбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҳбһҪбһҷбһҠбһҫбһҳбҹ’бһ”бһёбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбҹ” бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ“бһ№бһ„бһҡбҹҖбһ“бһ“бҹ…бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮбҹ”  

## бһҖбһ¶бһҡбһҺбҹӮбһ“бһ¶бҹҶ

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҮбҹҶбһ–бһјбһҖбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бҹ–

- бһҡбҹҖбһ“бһўбҹҶбһ–бһёбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ openai бһ“бһ·бһ„бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһӮбҹҶбһ“бһ·бһҸбһҡбһ”бһҹбҹӢбһңбһ¶бҹ”
- бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openai бҹ”
- бһҷбһӣбҹӢбһ–бһёбһҡбһ”бҹҖбһ”бһ”бҹ’бһҡбһҫбһӮбҹҶбһ“бһ·бһҸбһҠбһјбһ…бһҮбһ¶ prompt, бһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһ“бһ·бһ„ tokens бһҠбһҫбһҳбҹ’бһ”бһёбһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бҹ”

## бһӮбҹ„бһӣбһ”бҹҶбһҺбһ„бһҡбҹҖбһ“

бһ“бҹ…бһ…бһ»бһ„бһ”бһүбҹ’бһ…бһ”бҹӢбһҳбҹҒбһҡбҹҖбһ“бһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ“бһ№бһ„бһўбһ¶бһ…бҹ–

- бһ–бһ“бҹ’бһҷбһӣбҹӢбһҗбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһҮбһ¶бһўбҹ’бһңбһёбҹ”
- бһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ openaiбҹ”
- бһҖбҹҶбһҺбһҸбҹӢбһҖбһ¶бһҡбһҖбҹҶбһҺбһҸбҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһұбҹ’бһҷбһ”бҹ’бһҡбһҫ tokens бһ…бҹ’бһҡбһҫбһ“ бһ¬бһҸбһ·бһ… бһ“бһ·бһ„бһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҠбһјбһҡбһҹбһёбһҸбһ»бһҺбҹ’бһ бһ—бһ¶бһ– бһҠбһҫбһҳбҹ’бһ”бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһӣбһ‘бҹ’бһ’бһ•бһӣбһ•бҹ’бһӣбһ¶бһҹбҹӢбһ”бҹ’бһҸбһјбһҡбҹ”

## бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһҮбһ¶бһўбҹ’бһңбһё?

бһҷбҹүбһ¶бһ„бһ’бһҳбҹ’бһҳбһҸбһ¶ бһ“бҹ…бһ–бҹҒбһӣбһўбҹ’бһ“бһҖбһҹбһ„бҹӢбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё бһңбһ¶бһҳбһ¶бһ“бһўбҹҠбһёбһ“бһ’бһәбһ•бҹҒбһҹбһҳбһҪбһҷбһ”бҹ’бһҡбһ—бҹҒбһ‘бһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- Command-basedбҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё console бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶ бһ бһҫбһҷбһңбһ¶бһ”бҹҶбһ–бҹҒбһүбһ—бһ¶бһҡбһҖбһ·бһ…бҹ’бһ…бһҳбһҪбһҷбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ `git` бһҮбһ¶бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ–бһҪбһҖ Command-basedбҹ”
- User interface (UI)бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһҳбһ¶бһ“бһ…бҹҶбһҺбһ»бһ…бһ”бҹ’бһҡбһ‘бһ¶бһҖбҹӢбһ”бҹ’бһҡбһҫбһ”бҹ’бһҡбһ¶бһҹбҹӢбһҖбҹ’бһҡбһ¶бһ бҹ’бһңбһ·бһҖ (GUIs) бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ…бһ»бһ…бһ”бҹҠбһјбһҸбһ»бһ„ бһңбһ¶бһҷбһўбһҸбҹ’бһҗбһ”бһ‘ бһҮбҹ’бһҡбһҫбһҹбһҮбһҳбҹ’бһҡбһҫбһҹ бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”

### бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё Console бһ“бһ·бһ„ UI бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ

бһ”бҹ’бһҡбҹҖбһ”бһ’бҹҖбһ”бһңбһ¶бһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһё command-based бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бҹ–

- **бһңбһ¶бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢ**бҹ” бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһўбһ¶бһ…бһңбһ¶бһҷбһ”бһүбҹ’бһҮбһ¶бһҡбһҺбһ¶бһҳбһҪбһҷбһ”бһ¶бһ“бһ‘бҹҒ бһҸбҹӮбһҳбһ¶бһ“бһҸбҹӮбһ”бһүбҹ’бһҮбһ¶бһҠбҹӮбһӣбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһҸбҹӮбһ”бҹүбһ»бһҺбҹ’бһҺбҹ„бҹҮбҹ”
- **бһ•бҹ’бһўбҹӮбһҖбһӣбһҫбһ—бһ¶бһҹбһ¶**бҹ” бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҒбҹ’бһӣбҹҮбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“ бһ”бҹүбһ»бһ“бҹ’бһҸбҹӮбһӣбҹҶбһ“бһ¶бҹҶбһҠбһҫбһҳбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ“бҹҒбҹҮбһҸбҹ’бһҡбһјбһңбһ”бһ¶бһ“бһҹбһ„бҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ—бһ¶бһҹбһ¶бһҳбһҪбһҷбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢ бһ‘бҹ„бҹҮбһ”бһёбһ”бһ“бҹ’бһҗбҹӮбһҳбһҖбһ¶бһҡбһӮбһ¶бҹҶбһ‘бҹ’бһҡбһ—бһ¶бһҹбһ¶бһ”бһ“бҹ’бһҗбҹӮбһҳбһ”бһ¶бһ“бһҖбҹҸбһҠбҹ„бһҷбҹ”

### бһӮбһ»бһҺбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ“бҹғбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҠбһјбһ…бҹ’бһ“бҹҒбҹҮ бһҸбһҫбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҒбһ»бһҹбһ–бһёбһӮбҹ’бһ“бһ¶бһ”бҹӮбһ”бһҺбһ¶?

бһҖбҹ’бһ“бһ»бһ„бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘ бһўбҹ’бһ“бһҖбһҳбһ¶бһ“бһҹбһҳбһҸбҹ’бһҗбһ—бһ¶бһ–бһ”бһҸбҹӢбһ”бҹӮбһ“бһ…бҹ’бһҡбһҫбһ“бһҮбһ¶бһ„бһҳбһ»бһ“ бһҳбһ·бһ“бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһӣбһҫбһҖбҹ’бһҡбһ»бһҳбһ”бһүбҹ’бһҮбһ¶ бһ¬бһ—бһ¶бһҹбһ¶бһҖбһ¶бһҡбһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҖбҹӢбһӣбһ¶бһҖбҹӢбһ‘бҹҒбҹ” бһ•бҹ’бһ‘бһ¶бһӣбҹӢбһҒбҹ’бһӣбһҪбһ“ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ—бһ¶бһҹбһ¶бһ’бһҳбҹ’бһҳбһҮбһ¶бһҸбһ·бһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡвҖӢбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбҹ” бһӮбһ»бһҺбһ”бҹ’бһҡбһҷбҹ„бһҮбһ“бҹҚбһ•бҹ’бһҹбҹҒбһ„бһ‘бҹҖбһҸ бһӮбһәбһҠбҹ„бһҷбһҹбһ¶бһҡбһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ’бҹ’бһңбһҫбһўбһ“бҹ’бһҸбһҡбһҖбһҳбҹ’бһҳбһҮбһ¶бһҳбһҪбһҷбһ”бҹ’бһҡбһ—бһ–бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһҺбҹ’бһҠбһ»бҹҮбһ”бһҺбҹ’бһҠбһ¶бһӣбһӣбһҫбһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ’бҹҶбһҳбһҪбһҷ бһҒбһ»бһҹбһ–бһёбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бҹӮбһ”бһ”бҹ’бһҡбһ–бҹғбһҺбһёбһҠбҹӮбһӣбһўбһ¶бһ…бһҳбһ¶бһ“бһҖбҹҶбһҺбһҸбҹӢбһӣбһҫбһўбҹ’бһңбһёбһҠбҹӮбһӣбһҳбһ¶бһ“бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҳбһјбһӣбһҠбҹ’бһӢбһ¶бһ“бһ‘бһ·бһ“бҹ’бһ“бһ“бҹҗбһҷбҹ”  

### бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһўбһ¶бһ…бһҹбһ„бҹӢбһўбҹ’бһңбһёбһ”бһ¶бһ“бһҒбҹ’бһӣбҹҮбһҮбһ¶бһҳбһҪбһҷбһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘?

бһҳбһ¶бһ“бһҡбһҝбһ„бһҮбһ¶бһ…бҹ’бһҡбһҫбһ“бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ„бҹӢбһ”бһ¶бһ“бҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

- **Chatbot**бҹ” Chatbot бһҠбҹӮбһӣбһҶбҹ’бһӣбһҫбһҷбһҹбҹҶбһҺбһҪбһҡбһўбҹҶбһ–бһёбһ”бҹ’бһҡбһ’бһ¶бһ“бһ”бһ‘ бһҠбһјбһ…бһҮбһ¶ бһҖбҹ’бһҡбһ»бһҳбһ бҹҠбһ»бһ“ бһ“бһ·бһ„бһ•бһӣбһ·бһҸбһ•бһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҮбһ¶бһҮбһҳбҹ’бһҡбһҫбһҹбһӣбҹ’бһўбҹ”
- **бһҮбҹҶбһ“бһҪбһҷбһҖбһ¶бһҡ**бҹ” LLM бһӮбһәбһӣбҹ’бһўбһ“бҹ…бһҖбһ·бһ…бҹ’бһ…бһҖбһ¶бһҡбһҠбһјбһ…бһҮбһ¶ бһҹбһ„бҹ’бһҒбҹҒбһ”бһўбһҸбҹ’бһҗбһ”бһ‘ бһ‘бһ‘бһҪбһӣбһҷбһҖбһӮбһ“бҹ’бһӣбһ№бҹҮбһ–бһёбһўбһҸбҹ’бһҗбһ”бһ‘ бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһҠбһјбһ…бһҮбһ¶ resume бһ“бһ·бһ„бһ•бҹ’бһҹбҹҒбһ„бҹ—бһ‘бҹҖбһҸбҹ”
- **бһҮбҹҶбһ“бһҪбһҷбһҖбһјбһҠ**бҹ” бһўбһ¶бһҹбҹ’бһҡбҹҗбһҷбһӣбһҫбһҳбҹүбһјбһҠбҹӮбһӣбһ—бһ¶бһҹбһ¶бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҹбһ„бҹӢбһҮбҹҶбһ“бһҪбһҷбһҖбһјбһҠбһҳбһҪбһҷбһҠбҹӮбһӣбһҮбһҪбһҷбһўбҹ’бһ“бһҖбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ” бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚ бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһ•бһӣбһ·бһҸбһ•бһӣбһҠбһјбһ…бһҮбһ¶ GitHub Copilot бһ“бһ·бһ„ ChatGPT бһҠбһҫбһҳбҹ’бһ”бһёбһҮбһҪбһҷбһҹбһҡбһҹбҹҒбһҡбһҖбһјбһҠбҹ”  

## бһҸбһҫбһҒбҹ’бһүбһ»бҹҶбһ…бһ¶бһ”бҹӢбһ•бҹ’бһҸбһҫбһҳбһҠбҹ„бһҷбһҡбһ”бҹҖбһ”бһҺбһ¶?

бһ–бһ·бһҸбһ”бҹ’бһҡбһ¶бһҖбһҠ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһҡбһҖбһңбһ·бһ’бһёбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ”бһүбҹ’бһ…бһјбһӣбһҮбһ¶бһҳбһҪбһҷ LLM бһҠбҹӮбһӣбһ’бһҳбҹ’бһҳбһҸбһ¶бһҮбһ¶бһҡбһ”бҹҖбһ”бһ–бһёбһҡбһҠбһјбһ…бһҒбһ¶бһ„бһҖбҹ’бһҡбҹ„бһҳбҹ–

- бһ”бҹ’бһҡбһҫ APIбҹ” бһ“бҹ…бһ‘бһёбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҖбҹҶбһ–бһ»бһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһҖбһ¶бһҡбһҹбҹ’бһ“бһҫбһҹбһ»бҹҶбһңбҹӮбһ”бһҮбһ¶бһҳбһҪбһҷ prompt бһ бһҫбһҷбһ‘бһ‘бһҪбһӣбһўбһҸбҹ’бһҗбһ”бһ‘бһ”бһ„бҹ’бһҖбһҫбһҸбһҸбҹ’бһҡбһЎбһ”бҹӢбһҳбһҖбһңбһ·бһүбҹ”
- бһ”бҹ’бһҡбһҫбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҮбһҪбһҷбһҮбһ»бҹҶбһҹбҹҶбһҺбһҫ API бһ“бһ·бһ„бһ’бҹ’бһңбһҫбһІбҹ’бһҷбһңбһ¶бһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһ”бҹ’бһҡбһҫбҹ”

## бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ/SDKs

бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҡбһҪбһ…бһҡбһ¶бһӣбҹӢбһҹбһҳбҹ’бһҡбһ¶бһ”бҹӢбһ’бҹ’бһңбһҫбһҖбһ¶бһҡбһҮбһ¶бһҳбһҪбһҷ LLMs бһҠбһјбһ…бһҮбһ¶бҹ–

- **openai**бҹ” бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһ“бҹҒбҹҮбһ’бҹ’бһңбһҫбһұбҹ’бһҷбһ„бһ¶бһҷбһҹбҹ’бһҡбһҪбһӣбһҖбҹ’бһ“бһ»бһ„бһҖбһ¶бһҡбһҸбһ—бҹ’бһҮбһ¶бһ”бҹӢбһ‘бҹ…бһҳбҹүбһјбһҠбҹӮбһӣбһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ“бһ·бһ„бһ•бҹ’бһүбһҫ prompt бһ‘бҹ…бҹ”

бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһ–бһёбһ“бҹҒбҹҮ бһҳбһ¶бһ“бһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһҠбҹӮбһӣбһҠбҹҶбһҺбһҫбһҡбһҖбһ¶бһҡбһӣбһҫбһҹбһҖбһҳбҹ’бһҡбһ·бһҸбһҒбҹ’бһ–бһҹбҹӢбһҮбһ¶бһ„бһҠбһјбһ…бһҮбһ¶бҹ–

- **Langchain**бҹ” Langchain бһӮбһәбһӣбҹ’бһ”бһёбһ бһҫбһҷбһӮбһ¶бҹҶбһ‘бҹ’бһҡ Pythonбҹ”

## бһҖбһҳбҹ’бһҳбһңбһ·бһ’бһёбһҠбҹҶбһ”бһјбһ„бһ”бҹ’бһҡбһҫ openai

бһ“бҹ…бһ–бҹҒбһӣбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһ¶бһ”бҹӢбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷ `openai` бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҗбһ¶бһӮбһ“бҹ’бһӣбһ„ API key бһҺбһ¶бһҠбҹӮбһӣбһҸбҹ’бһҡбһјбһңбһ”бҹ’бһҡбһҫбҹ” бһҠбһҫбһҳбҹ’бһ”бһёбһҡбһҖ API key бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһҹбһјбһҳбһ‘бҹ…бһ”бһ“бҹ’бһ‘бһ¶бһҸбҹӢ "Keys and Endpoint" бһ“бҹғбһ’бһ“бһ’бһ¶бһ“ Azure OpenAI бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖ бһ бһҫбһҷбһ…бһҳбҹ’бһӣбһ„бһҸбһҳбҹ’бһӣбҹғ "Key 1"бҹ”

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

бһҘбһЎбһјбһңбһ“бҹҒбҹҮ бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ…бһҳбҹ’бһӣбһ„бһ–бҹҗбһҸбҹҢбһҳбһ¶бһ“бһ“бҹҒбҹҮбһҡбһҪбһ… бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһҺбҹӮбһ“бһ¶бҹҶбһ”бһҺбҹ’бһҺбһ¶бһӣбҹҗбһҷбһІбҹ’бһҷбһ”бҹ’бһҡбһҫбһңбһ¶бҹ”

> [!NOTE]
> бһ–бһ·бһҸбһ”бҹ’бһҡбһ¶бһҖбһҠбһҗбһ¶бһўбҹ’бһ“бһҖбһҸбҹ’бһҡбһјбһңбһ”бҹҶбһ”бҹӮбһҖ API key бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ–бһёбһҖбһјбһҠбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ’бҹ’бһңбһҫбһ”бһ¶бһ“бһҠбҹ„бһҷбһ”бҹ’бһҡбһҫ environment variablesбҹ”
> - бһҖбҹҶбһҺбһҸбҹӢ environment variable `AZURE_OPENAI_API_KEY` бһ‘бҹ… API key бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҜбһҖбһҹбһ¶бһҡ .envбҹ” бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһүбҹ’бһ…бһ”бҹӢбһӣбҹҶбһ бһ¶бһҸбҹӢбһҳбһ»бһ“бҹ—бһҡбһ”бһҹбҹӢбһңбһӮбҹ’бһӮбһ“бҹҒбҹҮбһ бһҫбһҷ бһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһҡбҹҖбһ”бһ…бҹҶбһҡбһҪбһ…бһҡбһ¶бһӣбҹӢбһ бһҫбһҷбҹ”


### бһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’ Azure

бһ”бҹ’бһҡбһҹбһ·бһ“бһ”бһҫбһўбҹ’бһ“бһҖбһ”бҹ’бһҡбһҫ Azure OpenAI бһ“бҹҒбҹҮбһҮбһ¶бһҡбһ”бҹҖбһ”бһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһҖбҹҶбһҺбһҸбҹӢбһҡбһ…бһ“бһ¶бһҹбһҳбҹ’бһ–бҹҗбһ“бҹ’бһ’бҹ” Responses API бһ•бҹ’бһҸбһӣбҹӢбһҹбҹҒбһңбһ¶бһҖбһҳбҹ’бһҳбһ–бһё Azure OpenAI **v1 endpoint**, бһҠбһјбһ…бһ“бҹҒбҹҮ бһҷбһҫбһ„бһҖбҹҶбһҺбһҸбҹӢ client `OpenAI` бһ‘бҹ… `<your-endpoint>/openai/v1/`бҹ–

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

бһҒбһ¶бһ„бһӣбһҫ бһҷбһҫбһ„бһҖбҹҶбһ–бһ»бһ„бһҖбҹҶбһҺбһҸбҹӢбҹ–

- `api_key` бҹ– бһ“бҹҒбҹҮбһҮбһ¶бһӮбһ“бҹ’бһӣбһ„ API key бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҠбҹӮбһӣбһ”бһ¶бһ“бһҡбһҖбһғбһҫбһүбһ“бҹ…бһҠбһ„ Azure Portalбҹ”
- `base_url` бҹ– бһ“бҹҒбҹҮбһҮбһ¶бһ…бҹҶбһҺбһ»бһ…бһ”бһүбҹ’бһ…бһ”бҹӢ Azure OpenAI бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбһҠбҹӮбһӣбһ—бҹ’бһҮбһ¶бһ”бҹӢбһҮбһ¶бһҳбһҪбһҷ `/openai/v1/`бҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһҡбһҖ endpoint бһ“бҹ… Azure Portal бһ“бҹ…бһҮбһ¶бһ”бҹӢбһҠбһ¶бһҖбҹӢ API key бһҡбһ”бһҹбҹӢбһўбҹ’бһ“бһҖбҹ” бһ”бҹ’бһҡбһҫ endpoint v1 бһҳбһ¶бһ“бһ“бҹҗбһҷбһҗбһ¶бһўбҹ’бһ“бһҖбһҳбһ·бһ“бһ…бһ¶бҹҶбһ”бһ¶бһ…бҹӢбһ”бһүбҹ’бһҮбһјбһ“ `api_version` бһ‘бҹҖбһҸбһ‘бҹҒбҹ”
- `deployment` бҹ– бһ“бҹҒбҹҮбһӮбһәбһҮбһ¶бһҲбҹ’бһҳбҹ„бҹҮбһ“бҹғбһҖбһҳбҹ’бһҡбһ·бһҸбһ•бҹ’бһҹбһ–бҹ’бһңбһ•бҹ’бһҹбһ¶бһҷбһҳбҹүбһјбһҠбҹӮбһӣбһҠбҹӮбһӣбһўбҹ’бһ“бһҖбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбһ“бҹ…бһ•бҹ’бһ‘бһ¶бҹҶбһ„ Foundryбҹ”

> [!NOTE]
> `os.environ` бһӮбһәбһҮбһ¶бһҹбҹҖбһңбһ—бҹ…бһ•бҹӮбһ“бһ‘бһёбһҠбҹӮбһӣбһўбһ¶бһ“ environment variablesбҹ” бһўбҹ’бһ“бһҖбһўбһ¶бһ…бһ”бҹ’бһҡбһҫбһңбһ¶бһўбһ¶бһ“ environment variables бһҠбһјбһ…бһҮбһ¶ `AZURE_OPENAI_API_KEY` бһ“бһ·бһ„ `AZURE_OPENAI_ENDPOINT`бҹ”

## бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘

бһҡбһ”бҹҖбһ”бһ”бһ„бҹ’бһҖбһҫбһҸбһўбһҸбҹ’бһҗбһ”бһ‘бһӮбһәбһ”бҹ’бһҡбһҫбһңбһ·бһ’бһё `responses.create`бҹ” бһ“бҹҒбҹҮбһҮбһ¶бһ§бһ‘бһ¶бһ бһҡбһҺбҹҚбҹ–

```python
prompt = "Complete the following: Once upon a time there was a"
```

бһ“бҹ…бһҖбҹ’бһ“бһ»бһ„бһҖбһјбһҠбһҒбһ¶бһ„бһӣбһҫ бһҷбһҫбһ„бһ”бһ„бҹ’бһҖбһҫбһҸбһңбһҸбҹ’бһҗбһ»бһ”бҹ’бһҡбһҸбһ·бһҖбһҳбҹ’бһҳ бһ бһҫбһҷбһ”бһүбҹ’бһҮбһјбһ“бһҳбҹүбһјбһҠбҹӮбһӣбһҠбҹӮбһӣбһҷбһҫбһ„бһ…бһ„бҹӢбһ”бҹ’бһҡбһҫ бһ“бһ·бһ„ promptбҹ” бһ”бһ“бҹ’бһ‘бһ¶бһ”бҹӢбһҳбһҖбһҷбһҫбһ„бһ”бҹ„бҹҮбһ–бһ»бһҳбҹ’бһ–бһўбһҸбҹ’бһҗбһ”бһ‘бһҠбҹӮбһӣбһ”бһ¶бһ“бһ”бһ„бҹ’бһҖбһҫбһҸбҹ”

### бһ…бһҳбҹ’бһӣбһҫбһҷбһҮбһ¶бһўбһ„бҹ’бһӮбһҡбһҖбҹ’бһҹ

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
