# បើកបរ្តខាងវិជ្ជាសំរាប់រចនាសំណួរ
ការសំរាប់រចនាសំណួរជាដំណើរការរចនារួចបន្ទាន់សំណួរសម្រាប់ភារកិច្ចចំណេះដឹងភាសាធម្មជាតិ។ វា​រួមមានការជ្រើសរើសសំណួរសមរម្យ ការតម្រូវប៉ារ៉ាម៉ែត្រ រួចវាយតម្លៃការប្រតិបត្តិ។ ការសំរាប់រចនាសំណួរមានសារៈសំខាន់ក្នុងការសម្រេចបានភាពត្រឹមត្រូវ និងប្រសិទ្ធភាពខ្ពស់ក្នុងគំរូ NLP។ នៅផ្នែកនេះ យើងនឹងស្វែងយល់ពីមូលដ្ឋាននៃការសំរាប់រចនាសំណួរដោយប្រើគំរូ OpenAI សម្រាប់ការស្វែងរក។


### លំហាត់ 1: ការបំបែកពាក្យ
ស្វែងយល់អំពីការបំបែកពាក្យដោយប្រើ tiktoken ដែលជាកម្មវិធីបំបែកពាក្យលឿនដែលទំនិញូក្រោមក្បួនឯករាជ្យពី OpenAI
សូមមើល [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) ដើម្បីឃើញឧទាហរណ៍បន្ថែមទៀត។


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### អនុស្សាវរីយ៍ 2: បញ្ជាក់ទុក្ខ API Key របស់ OpenAI

បើកដំណើរការកូដខាងក្រោម ដើម្បីបញ្ជាក់ថា endpoint របស់ OpenAI របស់អ្នកត្រូវបានកំណត់ឲ្យបានត្រឹមត្រូវ។ កូដនេះគ្រាន់តែសាកល្បងការបញ្ចូលសំណួរជាមូលដ្ឋានរួចទទួលបានការបញ្ចប់។ អត្ថបទបញ្ចូល `oh say can you see` គួរត្រូវបានបញ្ចប់ជា `by the dawn's early light..`


````python
# Uses the OpenAI client with the Responses API.
# See https://platform.openai.com/docs/api-reference/responses

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### ប练习 ៣៖ ការបង្កើតក្លែងក្លាយ
ស្វែងយល់ពីអ្វីដែលកើតឡើងនៅពេលអ្នកស្នើ LLM ដើម្បីត្រឡប់វិញនូវការបញ្ចប់សម្រាប់សំណួរអំពីប្រធានបទដែលប្រហែលជាមិនមាន ឬអំពីប្រធានបទដែលវាអាចមិនស្គាល់ព្រោះវាដែលនៅក្រៅពីឃ្លាំងឃ្លាំងទិន្នន័យដែលវាត្រូវបានបច្ចេកទេសជាមុន (កាន់តែថ្មី)។ មើលថាតើការឆ្លើយតបផ្លាស់ប្ដូរយ៉ាងដូចម្តេច ប្រសិនបើអ្នកសាកល្បងជាមួយសំណួរផ្សេង ឬម៉ូដែលផ្សេងទៀត។  


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### គណនា 4: អន្ដរជាធរណីមួយ

និងអថេរ "prompt" ដើម្បីផ្តល់ការណែនាំដែលទាក់ទងនឹងមាតិកាធំនោះ។

នៅទីនេះយើងសូមឱ្យម៉ូដែលសង្ខេបអត្ថបទសម្រាប់និស្សិតថ្នាក់ទីពីរ


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### កម្មវិធី ៥៖ បទបញ្ជាស្មុគស្មាញ  
សូមព្យាយាមសំណើដែលមានសារៈប្រយោជន៏ប្រព័ន្ធ អ្នកប្រើ និងជំនួយការជាសារ  
ប្រព័ន្ធកំណត់សេចក្ដីបរិបទជំនួយការ  
សារអ្នកប្រើ និងជំនួយការផ្តល់បរិបទការសន្ទនាច្រើនជំហាន  

សូមយកចិត្តទុកដាក់ថា បុគ្គលិកភាពរបស់ជំនួយការត្រូវបានកំណត់ជា "ពួកគេនិយាយអញ្ចឹង" ក្នុងបរិបទប្រព័ន្ធ។  
សូមព្យាយាមប្រើបុគ្គលិកភាពផ្សេងទៀត។ ឬសាកល្បងជួរប្រភេទសារបញ្ជូល/ចេញផ្សេងទៀត  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
