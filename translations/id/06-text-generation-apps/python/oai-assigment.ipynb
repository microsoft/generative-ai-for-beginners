## Apa itu aplikasi penghasil teks?

Biasanya saat Anda membangun sebuah aplikasi, aplikasi tersebut memiliki jenis antarmuka seperti berikut:

- Berbasis perintah. Aplikasi konsol adalah aplikasi biasa di mana Anda mengetik sebuah perintah dan aplikasi menjalankan tugas tersebut. Contohnya, `git` adalah aplikasi berbasis perintah.
- Antarmuka pengguna (UI). Beberapa aplikasi memiliki antarmuka pengguna grafis (GUI) di mana Anda klik tombol, memasukkan teks, memilih opsi dan lainnya.

### Aplikasi konsol dan UI memiliki keterbatasan

Bandingkan dengan aplikasi berbasis perintah di mana Anda mengetik sebuah perintah: 

- **Terbatas**. Anda tidak bisa mengetik sembarang perintah, hanya perintah-perintah yang didukung oleh aplikasi.
- **Spesifik bahasa**. Beberapa aplikasi mendukung banyak bahasa, tapi secara default aplikasi tersebut dibangun untuk bahasa tertentu, walaupun Anda bisa menambahkan dukungan bahasa lebih banyak. 

### Manfaat aplikasi penghasil teks

Lalu apa yang membedakan aplikasi penghasil teks?

Di aplikasi penghasil teks, Anda memiliki fleksibilitas lebih besar, Anda tidak terbatas pada sekumpulan perintah atau bahasa input tertentu. Sebaliknya, Anda bisa menggunakan bahasa alami untuk berinteraksi dengan aplikasi. Manfaat lain adalah karena Anda sudah berinteraksi dengan sumber data yang dilatih dengan korpus informasi besar, sedangkan aplikasi tradisional biasanya terbatas dengan data di database.

## Latihan - buat generator resep

Sekarang setelah kita menjalankan sebuah skenario, mari kita tulis kode yang sesuai dengan skenario yang ditunjukkan. Untuk melakukannya, ikuti langkah-langkah berikut:

1. Temukan variabel `prompt` dan ubah kodenya menjadi berikut ini:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Tingkatkan dengan menambahkan filter dan daftar belanja

Di sini kami menyarankan menambahkan prompt tambahan, tapi untuk itu berhasil, kita harus menambahkan hasil dari prompt pertama sebagai konteks ke prompt kedua.

    Temukan bagian dalam kode yang mencetak hasil dari prompt pertama dan tambahkan kode berikut di bawahnya:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # cetak respons
    print("Shopping list:")
    print(response.output_text)
    ```

    Perhatikan hal berikut:

    - Kita membuat prompt baru dengan menambahkan hasil dari prompt pertama ke prompt baru:

- **Sedikit catatan tentang panjang token**. Kita harus mempertimbangkan berapa banyak token yang kita perlukan untuk menghasilkan teks yang kita inginkan. Token membutuhkan biaya, jadi sebisa mungkin kita harus berhemat dalam penggunaan token. Misalnya, bisa kah kita menyusun prompt agar menggunakan token lebih sedikit?

   Untuk mengubah jumlah token yang digunakan, Anda bisa menggunakan parameter `max_output_tokens`. Misalnya, jika Anda ingin menggunakan 100 token, Anda bisa melakukan:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Bereksperimen dengan temperature**. Temperature adalah sesuatu yang belum kita sebutkan sejauh ini tetapi merupakan konteks penting untuk bagaimana program kita bekerja. Nilai temperature yang lebih tinggi menghasilkan keluaran yang lebih acak. Sebaliknya, nilai temperature yang lebih rendah menghasilkan keluaran yang lebih dapat diprediksi. Pertimbangkan apakah Anda menginginkan variasi dalam keluaran atau tidak.

   Untuk mengubah temperature, Anda dapat menggunakan parameter `temperature`. Misalnya, jika Anda ingin menggunakan temperature sebesar 0.5, Anda bisa melakukan:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Catatan, semakin dekat ke 1.0, keluaran akan semakin bervariasi.
