#### ప్రాంప్ట్‌లు సృష్టించడానికి మూడు ప్రాథమిక మార్గదర్శకాలు ఉంటాయి:

**చూపించండి మరియు చెప్పండి**. మీరు ఏమి కోరుతున్నారో స్పష్టంగా చెప్పండి, అది సూచనలు, ఉదాహరణలు లేదా రెండింటి కలయిక ద్వారా కావచ్చు. మీరు మోడల్‌ను అంశాల జాబితాను వరుసగా ఆర్డర్ చేయాలని లేదా పేరాగ్రాఫ్‌ను భావోద్వేగం ప్రకారం వర్గీకరించాలని కోరుకుంటే, అది ఇదే అని చూపించండి.

**మంచి தரவு అందించండి**. మీరు క్లాసిఫయర్‌ను నిర్మించాలని లేదా మోడల్‌ను ఒక నమూనాను అనుసరించమని కోరుకుంటే, సరిపడిన ఉదాహరణలు ఉండాలని చూసుకోండి. మీ ఉదాహరణలను పరిశీలించండి — మోడల్ సాధారణంగా ప్రాథమిక సరైన స్పెల్లింగ్ తప్పుల ద్వారా కూడా సమాధానం ఇవ్వగలదు, కానీ ఇది ఉద్దేశపூர్వకమని భావించి సమాధానానికి ప్రభావం కలిగించవచ్చు.

**మీ సెట్టింగ్లను తనిఖీ చేయండి.** టెంపరేచర్ మరియు టాప్_p సెట్టింగ్లు మోడల్ ప్రతిస్పందన సృష్టించడంలో ఎంత నిర్దిష్టంగా ఉన్నదో నియంత్రిస్తాయి. మీరు ఒక్కటే సరైన జవాబు ఉన్న ప్రతిస్పందన కోరుతున్నట్లయితే, ఈ సెట్టింగ్‌లను తక్కువగా ఉంచాలి. మీరు విభిన్నమైన ప్రతిస్పందనలు ఆశిస్తే, ఈ సెట్టింగ్లను ఎక్కువగా ఉంచవచ్చు. ఈ సెట్టింగ్లతో ప్రజలు చేసే ప్రధాన తప్పు వాటిని "మేధా" లేదా "సృజనాత్మకత" নিয়ంత్రణలు అని భావించడమే.

### 5. సమర్పించండి!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
