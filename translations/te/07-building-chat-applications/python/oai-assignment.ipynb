#### పెద్ద భాషా నమూనాను ఎలా నియంత్రించాలి  
"పెద్ద భాషా నమూనాకు ఇచ్చే ఇన్పుట్‌లలో, అత్యంత ప్రభావవంతమైనది టెక్స్ట్ ప్రాంప్ట్(1) మాత్రమే.

పెద్ద భాషా నమూనాలను ఉత్పత్తి చేయించడానికి కొన్ని మార్గాలు ఉన్నాయి:

సూచన: మీరు కావాలనుకున్నదాన్ని నమూనాకు చెప్పండి
పూర్తి చేయడం: మీరు కావలసినదాన్ని మొదటివారిని పూర్తి చేయమని నమూనాను ప్రేరేపించండి
ప్రదర్శన: మీరు కావలసినదాన్ని నమూనాకు చూపండి, ఈ రెండు లో ఒకటి లేదా రెండింటిని ఉపయోగించండి:
ప్రాంప్ట్‌లో కొన్ని ఉదాహరణలు
ఫైన్-ట్యూనింగ్ శిక్షణ డేటాసెట్టులో నూరు లేదా వేలేకొన్ని ఉదాహరణలు"



#### ప్రాంప్ట్‌లు సృష్టించడానికి మూడు ప్రాథమిక మార్గదర్శకాలు ఉన్నాయి:

**చూపించు మరియు చెప్పు**. మీరు కావాలనుకున్నదాన్ని సూచనలు, ఉదాహరణలు లేదా రెండింటి కలయిక ద్వారా స్పష్టంగా తెలియజేయండి. మీరు నమూనాను అంశాల జాబితాను సార్వత్రిక క్రమంలో ర్యాంక్ చేయమని లేదా పేరాగ్రాఫ్‌ను భావోద్వేగం ద్వారా వర్గీకరించమని కోరితే, అది ఒక్కసారిగా కావాలని చూపించండి.

**నాణ్యమైన డేటాను అందించండి**. మీరు వర్గీకర్తను నిర్మించడానికి ప్రయత్నిస్తున్నట్లయితే లేదా నమూనాను ఒక నమూనాను అనుసరించమని కోరుతున్నట్లయితే, సరిపడిన ఉదాహరణలు ఉండేలా చూసుకోండి. మీ ఉదాహరణలను సవివరంగా పరిక్షించండి — నమూనా సాధారణంగా ప్రాథమిక అక్షర దోషాలను గుర్తించి సమాధానం ఇస్తుంది, కానీ ఇది ఉద్దేశపూర్వకమేనని అనుకుని సమాధానంపై ప్రభావం చూపించవచ్చు.

**మీ సెట్టింగ్స్‌ను తనిఖీ చేయండి.** టెంపరేచర్ మరియు టాప్_p సెట్టింగ్స్ నమూనా సమాధానాన్ని ఎంత నిర్దేశించదగినదిగా ఉత్పత్తి చేస్తుందో నియంత్రిస్తాయి. ఒకే సరైన జవాబు ఉన్న సమాధానాన్ని కోరుకుంటే, ఈ సెట్టింగ్స్‌ని తక్కువగా ఉంచండి. విభిన్నమైన సమాధానాలు కావాలనుకుంటే, ఇవి ఎక్కువగా ఉండాలి. ఈ సెట్టింగ్స్ పట్ల జనరల్ తప్పు భావన ఈ సెట్టింగ్స్ "పట్టుదల" లేదా "సృష్టి" నియంత్రణలు అని అనుకోవడం.

### 5. సమర్పించండి!


```python
# Create your first prompt
text_prompt = "Should oxford commas always be used?"

response = client.responses.create(
  model=model,
  input = [{"role":"system", "content":"You are a helpful assistant."},
               {"role":"user","content":text_prompt},],
  store=False,)

response.output_text
```
