## ఫంక్షన్ కాల్స్ అర్థం చేసుకోవడం

ఈ పాఠం కోసం, మనం ఒక ఎడ్యుకేషన్ స్టార్టప్ కోసం ఒక ఫీచర్ తయారుచేయాలని ఉంది, ఇది వినియోగదారులు టెక్నికల్ కోర్సులు కనుగొనడానికి చాట్‌బాట్ ఉపయోగించడానికి అనుమతిస్తుంది. వారి నైపుణ్య స్థాయి, ప్రస్తుత పాత్ర మరియు ఆసక్తి ఉన్న సాంకేతికతను ఆధారంగా కోర్సులను సూచించబోతున్నాము.

దీన్ని పూర్తి చేయడానికి, మనము ఈ సమ్మేళనాన్ని ఉపయోగిస్తాము:
 - వినియోగదారుకి చాట్ అనుభవం సృష్టించడానికి `OpenAI`
 - వినియోగదారుల అభ్యర్థన ఆధారంగా కోర్సులు కనుగొనడంలో సహాయం చేయడానికి `Microsoft Learn Catalog API`
 - వినియోగదారుని ప్రశ్నను తీసుకుని API అభ్యర్థన చేయడానికి ఫంక్షన్ కాలింగ్

ముందుగా ప్రారంభించడానికి, మనం ఎందుకు ఫంక్షన్ కాలింగ్‌ని ఉపయోగించాలో చూద్దాం:


### ఫంక్షన్ కాలింగ్ ఎందుకు  

మీరు ఈ కోర్సులో ఏదైనా ఇతర పాఠం పూర్తి చేసినట్లయితే, మీరు పెద్ద లాంగ్వేజ్ మోడల్స్ (LLMs) ఉపయోగించే శక్తిని అవగాహన చేసుకున్నారనుకుంటాను. ఆశిస్తుంది మీరు వాటి కొన్ని పరిమితులను కూడా చూస్తున్నారని.  

ఫంక్షన్ కాలింగ్ అనేది OpenAI సర్వీస్ యొక్క ఒక ఫీచర్, ఇది క్రింది సవాళ్లను ఎదుర్కొనడానికి రూపొందించబడింది:  

అసమ్మతమైన స్పందన ఫార్మాటింగ్:  
- ఫంక్షన్ కాలింగ్ ముందు, పెద్ద భాషా మోడల్ నుండి స్పందనలు నిర్మాణరహితంగా మరియు అసమ్మతంగా ఉండేవి. అభివృద్ధికర్తలు ప్రతి మార్పులో వేరుగా సంక్లిష్టమైన ధృవీకరణ కోడ్ రాయవలసి ఉండేది.  

బాహ్య డేటాతో పరిమిత అనుసంధానం:  
- ఈ ఫీచర్ రావడానికి ముందు, చాట్ సందర్భంలో అనువర్తనం ఇతర భాగాల నుండి డేటాను జోడించడం కష్టం అయ్యేది.  

స్పందన ఫార్మాట్లను స్తandirడైజ్ చేసి, బాహ్య డేటాతో సులభంగా అనుసంధానాన్ని సాధించడం ద్వారా, ఫంక్షన్ కాలింగ్ అభివృద్ధిని సులభతరం చేస్తుంది మరియు అదనపు ధృవీకరణ లాజిక్ అవసరాన్ని తగ్గిస్తుంది.  

వినియోగదారులు "స్టాక్‌హోమ్‌లో ప్రస్తుత వాతావరణం ఏంటి?" అనే ప్రశ్నలకు సమాధానాలు పొందలేకపోయారు. ఇది ఎందుకంటే మోడల్స్ శిక్షణ పొందిన డేటా సమయానికి పరిమితమైనవి.  

ఈ సమస్యను ప్రదర్శించే క్రింది ఉదాహరణను చూద్దాం:  

మనం సరైన కోర్సును సూచించేందుకు విద్యార్థుల డేటాబేస్ తయారు చేయాలనుకోండి. క్రింద రెండు విద్యార్థుల వివరణలు ఉన్నాయి, వీటి డేటా చాలా సమానంగా ఉంది.  


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


మేము డేటాను విశ్లేషించడానికి దీన్ని LLM కు పంపాలనుకుంటున్నాము. ఇది తరువాత మా అప్లికేషన్‌లో దీనిని APIకి పంపడానికి లేదా డేటాబేస్‌లో నిల్వ చేయడానికి ఉపయోగించవచ్చు.

మనం రెండు సమానమైన ప్రాంప్ట్లను సృష్టిద్దాం, వాటిద్వారా మనం LLMకి మనం ఆసక్తి కలిగిన సమాచారం ఏమిటి అనేది తెలిపేలా సూచిస్తాము:


మన ఉత్పత్తికి ముఖ్యమైన భాగాలను విశ్లేషించడానికి దీన్ని ఒక LLM కు పంపించాలనుకుంటున్నాము. అందువల్ల LLM ని మార్గదర్శనం చేసేందుకు ఒకే రకమైన రెండు ప్రశ్నలను సృష్టించవచ్చు: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ఈ రెండు ప్రాంప్ట్‌లను సృష్టించిన తర్వాత, వాటిని `client.responses.create` ఉపయోగించి LLM కి పంపిస్తాము. మేము ప్రాంప్ట్‌ను `input` వేరియబుల్‌లో సుంచుతాము మరియు పాత్రను `user` గా కేటాయిస్తాము. ఇది యూజర్ నుండి చాట్‌బాట్‌కు సందేశం రాయబడుతోందని అనుకరించడానికి చేయబడింది.



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"
```

ఇప్పుడు మనం రెండు అభ్యర్థనలను LLM కు పంపించి, మనం పొందే ప్రతిస్పందనను పరిశీలించవచ్చు.


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ప్రాంప్ట్లు ఒకటే అయినప్పటికీ మరియు వివరణలు సమానమైనప్పటికీ, మేము `Grades` లక్షణం యొక్క భిన్న ఫార్మాట్స్ పొందవచ్చు.

పై సెల్‌ను ఒకাধিক సార్లు నడుపితే, ఫార్మాట్ `3.7` లేదా `3.7 GPA` కావచ్చు.

ఇది జరిగిన కారణం LLM రాయబడిన ప్రాంప్ట్ రూపంలో అసంరచిత డేటాను తీసుకొని, అసంరచిత డేటాను తిరిగి ఇస్తుంది. మేము ఒక సుస్థిర ఫార్మాట్ అవసరమై ఉంటుంది, తద్వారా ఈ డేటాను నిల్వ చేయడం లేదా ఉపయోగించడం సమయంలో ఏమి ఆశించాలో అర్థం అవుతుంది.

ఫంక్షనల్ కాలింగ్ ఉపయోగించడం ద్వారా, మేము సుస్థిర డేటాను తిరిగి పొందగలుగుతాము. ఫంక్షన్ కాలింగ్ ఉపయోగించినప్పుడు, LLM అసలు ఏ ఫంక్షన్‌లను పిలవదో లేదా నడుపదో చేయదు. అంతకంటే, ఆ విభాగానికి అనుసరించాల్సిన నిర్మాణాన్ని మేము సృష్టిస్తాము. మా అప్లికేషన్లలో ఏ ఫంక్షన్ నడపాలో తెలియడానికి ఆ నిర్మాణిత ప్రతిస్పందనలను వాడతాము.

**API లేదా డేటాబేస్ క్వెరీస్ సృష్టించడం**  
యూజర్లు సహజ భాషలో సమాచారాన్ని ప్రశ్నించి, దాన్ని ఫార్మాట్ చేయబడిన క్వెరీ లేదా API అభ్యర్థనగా మార్చవచ్చు. దీని ఒక ఉదాహరణగా, ఒక ఉపాధ్యాయుడు "చివరి అసైన్‌మెంట్ పూర్తి చేసిన విద్యార్థులు ఎవరు" అని అడుగుతూ `get_completed(student_name: string, assignment: int, current_status: string)` అనే ఫంక్షన్‌ను కాల్ చేయవచ్చు  


**రచనాత్మక డేటా సృష్టించడం**  
యూజర్లు ఒక టెక్స్ట్ బ్లాక్ లేదా CSV తీసుకుని LLM ను ఉపయోగించి ముఖ్యమైన సమాచారాన్ని పొందవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి శాంతి ఒప్పందాలపై వికీపీడియా ఆర్టికల్‌ను AI ఫ్లాష్ కార్డులుగా మార్చవచ్చు. ఇది `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` అనే ఫంక్షన్ ద్వారా చేయవచ్చు  


## 2. మీ మొదటి ఫంక్షన్ కాల్‌ను సృష్టించడం  

ఒక ఫంక్షన్ కాల్‌ను సృష్టించే ప్రక్రియలో 3 ప్రధాన దశలు ఉంటాయి:  
1. మీ ఫంక్షన్లు మరియు ఒక వినియోగదారుని సందేశం జాబితాతో Chat Completions API ను కాల్ చేయడం  
2. ఒక చర్యను నిర్వహించడానికి మోడల్ ప్రతిస్పందనను చదవడం అంటే ఒక ఫంక్షన్ లేదా API కాల్‌ను నడపడం  
3. వినియోగదారునికి ప్రతిస్పందన సృష్టించడానికి ఆ సమాచారాన్ని ఉపయోగించేందుకు మీ ఫంక్షన్ నుండి వచ్చిన ప్రతిస్పందనతో మరోసారి Chat Completions API ని కాల్ చేయడం.  


![ఫంక్షన్ కాల్ పరిప్రవాహం](../../../../translated_images/te/LLM-Flow.3285ed8caf4796d7.webp)


### ఫంక్షన్ కాల్ యొక్క భాగాలు

#### వినియోగదారుల ఇన్‌పుట్

తొలి քայլ ఉంది వినియోగదారు సందేశాన్ని సృష్టించడం. ఇది ఒక టెక్స్ట్ ఇన్‌పుట్ విలువను తీసుకొని డైనమిక్గా కేటాయించవచ్చు లేదా మీరు ఇక్కడ విలువను కేటాయించవచ్చు. మీరు Chat Completions APIతో మొదటిసారి పని చేస్తుంటే, మేము సందేశం యొక్క `role` మరియు `content` ని నిర్వచించాలి.

`role` అనేది `system` (నియమాలు సృష్టించడం), `assistant` (మోడల్) లేదా `user` (అంతిమ వినియోగదారు) లో ఏదైనా కావచ్చు. ఫంక్షన్ కాలింగ్ కోసం, మేము దీన్ని `user` గా కేటాయించి ఒక ఉదాహరణ ప్రశ్నను ఇస్తాము.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ఫంక్షన్లు సృష్టించడం. 

తర్వాత మేము ఒక ఫంక్షన్ మరియు ఆ ఫంక్షన్ యొక్క పారామితులను నిర్వచిస్తాము. ఇక్కడ `search_courses` అనే ఒక్క ఫంక్షన్ ను మాత్రమే ఉపయోగిస్తాము, కానీ మీరు అనేక ఫంక్షన్లు సృష్టించవచ్చు.

**ముఖ్యమైనది** : ఫంక్షన్లు LLM కి సిస్టం సందేశాలలో చేర్చబడి, అందుబాటులో ఉన్న టోకెన్ల పరిమితిలో కూడా ఉండవచ్చు. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**నిర్వచనలు** 

ఫంక్షన్ నిర్వచన నిర్మాణం అనేక స్థాయిలను కలిగి ఉంటుంది, ప్రతి ఒక్కదానికి అసలైన గుణాలు ఉంటాయి. ఇక్కడ ఆ లోపలనున్న నిర్మాణం వివరించబడింది:

**తొలి స్థాయి ఫంక్షన్ గుణాలు:**

`name` - మనం పిలవదలచిన ఫంక్షన్ పేరు. 

`description` - ఫంక్షన్ ఎలా పనిచేస్తుందో యొక్క వివరణ. ఇక్కడ స్పష్టంగా మరియు నిర్దిష్టంగా ఉండటం ముఖ్యం 

`parameters` - మోడల్ ప్రతిస్పందనలో ఉత్పత్తి చేయదలచిన విలువలు మరియు ఫార్మాట్ జాబితా 

**పరామితుల آب్జెక్టు గుణాలు:**

`type` - పరామితుల آب్జెక్టు యొక్క డేటా రకం (సాధారణంగా "object")

`properties` - మోడల్ ప్రతిస్పందన కోసం ఉపయోగించే ప్రత్యేక విలువల జాబితా 

**ప్రత్యేక పరామితి గుణాలు:**

`name` - ప్రాపర్టీ కీ ద్వారా స్పష్టంగా నిర్వచింపబడినది (ఉదా: "role", "product", "level")

`type` - ఈ ప్రత్యేక పరామితి యొక్క డేటా రకం (ఉదా: "string", "number", "boolean") 

`description` - ఈ ప్రత్యేక పరామితి యొక్క వివరణ 

**ఐచ్ఛిక గుణాలు:**

`required` - ఫంక్షన్ కాల్ పూర్తి చేయడానికి కావలసిన పరామితుల జాబితా 


### ఫంక్షన్ కాల్ చేయడం  
ఒక ఫంక్షన్ నిర్వచించిన తర్వాత, ఇప్పుడు దాని కాల్‌ను Chat Completion API కి చేర్చాలి. మేము ఇదిని `functions` ను రిక్వెస్ట్‌ లో చేర్చటం ద్వారా చేస్తాం. ఈ సందర్భంలో `functions=functions`.  

`function_call` ని `auto` గా సెట్ చేసే ఎంపిక కూడా ఉంది. ఇది అర్థం, మేము మనం నిర్ణయించుకోవడానికంటే వినియోగదారుడు సందేశంపై ఆధారపడి ఏ ఫంక్షన్ కాల్ కావాలో LLM నిర్ణయిస్తుంది.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ఇప్పుడు మనం ప్రతిస్పందనను చూడండి మరియు అది ఎలా రూపకల్పన చేయబడిందో చూడండి: 

{
  "role": "assistant",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

మీరు చూడగలరు ఫంక్షన్ పేరు పిలవబడింది మరియు యూజర్ సందేశం నుండి, LLM ఆర్గుమెంట్లకు సరిపోయే డేటాను కనుగొనగలిగింది. 


## 3. ఫంక్షన్ కాల్స్‌ను అనువర్తనంలో ఇంటిగ్రేట్ చేయడం. 


LLM నుండి ఫార్మాట్ చేసిన స్పందనను పరీక్షించిన తర్వాత, ఇప్పుడు దీన్ని ఒక అనువర్తనంలో ఇంటిగ్రేట్ చేయవచ్చు. 

### ఫ్లోను నిర్వహించడం 

దీన్ని మా అనువర్తనంలో ఇంటిగ్రేట్ చేసేందుకు, క్రింది దశలను అనుసరించండి: 

మొదట, OpenAI సర్వీసులకు కాల్ చేసి స్పందన సందేశాన్ని `response_message` అనే వేరియబుల్‌లో నిల్వ చేయండి. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ఇప్పుడు మేము కోర్సుల జాబితాను పొందడానికి Microsoft Learn APIని పిలిచే ఫంక్షన్‌ను నిర్వచించబోతున్నాము: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ఉత్తమ ఆచరణగా, తర్వాత మోడల్ ఒక ఫంక్షన్‌ని కాల్ చేయదలచుకొన్నదో లేదో చూస్తాము. ఆ తరువాత, అందుబాటులో ఉన్న ఫంక్షన్లలో ఒకటిని సృష్టించి, కాల్ చేయబడుతున్న ఫంక్షన్‌కు సరిపోతుందో లేదో సరిపోలుస్తాము. 
ఆ తర్వాత, ఆ ఫంక్షన్ యొక్క_ARGUMENTS_ ను తీసుకుని వాటిని LLM నుండి వచ్చిన_arguments_ తో మ్యాప్ చేస్తాము.

చివరిగా, ఫంక్షన్ కాల్ సందేశం మరియు `search_courses` సందేశం ద్వారా తిరిగి వచ్చిన విలువలను జోడిస్తాము. ఇది LLM కి 
సహజ భాష ఉపయోగించి వినియోగదారుని ప్రతిస్పందించడానికి అవసరమైన అన్ని సమాచారం అందిస్తుంది.
