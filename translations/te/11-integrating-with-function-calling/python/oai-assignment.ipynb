## ఫంక్షన్ కాల్స్‌ను అర్థం చేసుకోవడం

ఈ పాఠం కోసం, మన విద్యా స్టార్టప్ కోసం ఒక ఫీచర్‌ని సృష్టించాలనుకుంటాము, ఇది వినియోగదారులు టెక్నికల్ కోర్సులను కనుగొనటానికి చాట్‌బాట్ ఉపయోగించడానికి అనుమతిస్తుంది. వారి నైపుణ్య స్థాయి, ప్రస్తుత పాత్ర మరియు ఆసక్తి టెక్నాలజీకి సరిపోయే కోర్సులను మేము సిఫార్సు చేస్తాము.

దీనిని పూర్తి చేయడానికి మేము కింది వాటి మిశ్రమాన్ని ఉపయోగించనున్నాము:
 - వినియోగదారులకు చాట్ అనుభవాన్ని సృష్టించడానికి `OpenAI`
 - వినియోగదారుల అభ్యర్థనల ఆధారంగా కోర్సులు కనుగొనటానికి `Microsoft Learn Catalog API`
 - వినియోగదారుల ప్రశ్నను తీసుకొని API అభ్యర్థన చేయడానికి ఒక ఫంక్షన్‌కు పంపడానికి `Function Calling`

ప్రారంభించడానికి, మొదటగా మనం ఎందుకు ఫంక్షన్ కాలింగ్‌ను ఉపయోగించాలి అనేదాన్ని పరిశీలిద్దాం:


### ఫంక్షన్ కాల్ చేయడంవలన ఏమిటి 

మీరు ఈ కోర్సులో ఎటువంటి ఇతర పాఠాన్ని పూర్తి చేసాక ఉంటే, మీరు పెద్ద భాషా మోడల్స్ (LLMs) వాడటం యొక్క శక్తిని అర్థం చేసుకుంటారు. ఆశిస్తున్నాము మీరు వాటి కొంత పరిమితులను కూడా చూడగలరు. 

ఫంక్షన్ కాలింగ్ అనేది OpenAI సర్వీస్ యొక్క ఒక ఫీచర్, ఇది క్రింది సవాళ్లను పరిష్కరించడానికి రూపొందించబడింది:

అసమంజసమైన స్పందన ఫార్మాటింగ్:
- ఫంక్షన్ కాలింగ్ ముందు, పెద్ద భాషా మోడల్ నుండి వచ్చిన స్పందనలు అసంఘటితంగా మరియు అసమంజసంగా ఉండేవి. డెవలపర్లు ప్రతీ అవుట్పుట్ వేరియేషన్‌ని నిర్వహించడానికి క్లిష్టమైన ధృవీకరణ కోడ్ రాయాల్సి వచ్చేది.

బాహ్య డేటాతో పరిమిత ఇంటిగ్రేషన్:
- ఈ ఫీచర్ ముందు, చాట్ సందర్భంలో అనువర్తనం యొక్క ఇతర భాగాల డేటాను చేర్చడం కష్టమయ్యేది.

స్పందన ఫార్మాట్లను ప్రామాణీకరించి, బాహ్య డేటాతో సులభంగా ఇంటిగ్రేట్ చేయగలిగే విధంగా చేయడం ద్వారా, ఫంక్షన్ కాలింగ్ అభివృద్ధిని సులభతరం చేసి, అదనపు ధృవీకరణ లాజిక్ అవసరాన్ని తగ్గిస్తుంది.

వినియోగదారులు "స్టాక్‌హోమ్‌లో ప్రస్తుతం వాతావరణం ఎలా ఉంది?" అని ప్రశ్న అడగలేకపోయారు. ఇది ఎందుకంటే మోడల్స్ శిక్షణ పొందిన డేటా సమయానికి పరిమితం అయ్యాయి. 

ఈ సమస్యను వివరించే క్రింది ఉదాహరణను చూద్దాం: 

మనం విద్యార్థుల డేటాబేస్‌ను సృష్టించి, సరైన కోర్సును సూచించాలనుకుందాం. క్రింద మనకు రెండు విద్యార్థుల వివరణలు ఉన్నాయి, వీటి డేటా చాలా సమానంగా ఉంటుంది.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


మేము ఈ డేటాను పార్స్ చేయడానికి LLM కి పంపాలని అనుకుంటున్నాము. తర్వాత దీన్ని మా అప్లికేషన్‌లో API కి పంపించడానికి లేదా డేటాబేస్‌లో నిల్వ చేసుకోవడానికి ఉపయోగించవచ్చు.

మనం రెండు సాదృశ్య ప్రాంప్టులను తయారుచేద్దాం, వాటిలో మనం LLM కు మనకు కావలసిన సమాచారాన్ని సూచిస్తాము:


మేము దీన్ని ఒక LLM కు పంపాలి మా ఉత్పత్తికి ముఖ్యమైన భాగాలను విశ్లేషించడానికి. కాబట్టి మేము LLM కు ఆదేశించడానికి రెండు సమానమైన ప్రాంప్ట్‌లు సృష్టించవచ్చు: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ఈ రెండు ప్రాంప్టులను సృష్టించిన తర్వాత, మేము వాటిని `client.responses.create` ఉపయోగించి LLM కి పంపుతాము.  మేము ప్రాంప్ట్‌ను `input` వేరియబుల్‌లో నిల్వచేస్తాము మరియు పాత్రను `user` గా నియమిస్తాము. ఇది యూజర్ నుండి చాట్‌బాట్‌కు పంపే సందేశాన్ని అనుకరించడానికి. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"
```

ఇప్పుడు మేము రెండు అభ్యర్థనలను LLM కు పంపి మేము పొందే ప్రతిస్పందనను పరిశీలించవచ్చు. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ప్రాంప్ట్‌లు ఇదే అయినప్పటికీ, వివరణలు సమానమైనప్పటికీ, `Grades` ప్రాపర్టీ వివిధ ఫార్మాట్లలో రావచ్చు.

పై సెల్‌ను అనేక సార్లు నడిపితే, ఫార్మాట్ `3.7` లేదా `3.7 GPA` కావచ్చు.

ఇది ఏమి అంటే, LLM రాయబడిన ప్రాంప్ట్ రూపంలో అసంరచిత డేటాను తీసుకుంటుంది మరియు అలాగే అసంరచిత డేటాను మళ్లీ ఇవ్వుతుంది. మనం ఒక నిర్మిత ఫార్మాట్ అవసరం ఎందుకంటే డేటాను భద్రపరచేప్పుడు లేదా ఉపయోగించే సమయంలో ఏం వేచి చూడాలో తెలుసుకోవాలి.

ఫంక్షనల్ కాలింగ్ ఉపయోగించడం ద్వారా, మనం నిర్మిత డేటాను తిరిగి పొందుతామనే నిశ్చితం చేసుకోవచ్చు. ఫంక్షన్ కాలింగ్ ఉపయోగించే సమయంలో, LLM వాస్తవంగా ఎలాంటి ఫంక్షన్లు పిలవదు లేదా నడిపించదు. దీని బదులు, మనం LLM తన ప్రతిస్పందనల కోసం అనుసరించవలసిన ఒక నిర్మాణాన్ని సృష్టిస్తాం. ఆ తర్వాత ఆ నిర్మిత ప్రతిస్పందనలను మన అప్లికేషన్లలో ఏ ఫంక్షన్ నడపాలో తెలుసుకోవడానికి ఉపయోగిస్తాం.

### ఫంక్షన్ కాల్స్ ఉపయోగించే సందర్భాలు 

**బాహ్య టూల్స్‌ను కాల్ చేయడం** 
చాట్‌బోట్లు వినియోగదారుల ప్రశ్నలకు సమాధానాలు ఇవ్వడంలో అద్భుతంగా ఉంటాయి. ఫంక్షన్ కాలింగ్ ఉపయోగించి, చాట్‌బోట్లు వినియోగదారుల సందేశాలను ఉపయోగించి కొన్ని పనులను పూర్తిచేయవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి "ఈ विषयంపై నాకు ఇంకొంచెం సహాయం అవసరం అని నా అధ్యాపകർకి ఇ మెయిల్ పంపించు" అని చాట్‌బాట్‌ను అడగవచ్చు. దీని కోసం `send_email(to: string, body: string)` అనే ఫంక్షన్‌ కాల్ చేయవచ్చు


**API లేదా డేటాబేస్ క్వెరీలను సృష్టించడం**
వినియోగదారులు సహజ ভাষా ఉపయోగించి సమాచారాన్ని కనుగొనవచ్చు, అది ఫార్మాటెడ్ క్వెరీ లేదా API అభ్యర్థనగా మారుతుంది. ఉదాహరణకి, ఒక టీచర్ "గత అసైన్‌మెంట్ పూర్తి చేసిన విద్యార్థులు ఎవరుగురు?" అని అడిగితే, `get_completed(student_name: string, assignment: int, current_status: string)` అనే ఫంక్షన్‌ను కాల్ చేయవచ్చు


**సంరచనాత్మక డేటాను సృష్టించడం**
వినియోగదారులు ఒక టెక్స్ట్ బ్లాక్ లేదా CSV తీసుకుని LLM ఉపయోగించి ముఖ్యమైన సమాచారాన్ని గుర్తించవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి శాంతి ఒప్పందాల గురించి వికీపీడియా వ్యాసాన్ని AI ఫ్లాష్ కార్డులుగా మార్చవచ్చు. ఇది `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` అనే ఫంక్షన్‌ను ఉపయోగించి చేయవచ్చు


## 2. మీ మొదటి ఫంక్షన్ కాల్ సృష్టించడం 

ఫంక్షన్ కాల్ సృష్టించే ప్రక్రియలో 3 ప్రధాన దశలు ఉంటాయి: 
1. మీ ఫంక్షన్ల జాబితా మరియు వినియోగదారు సందేశంతో చాట్ కంప్లీషన్స్ APIని కాల్ చేయడం 
2. ఒక చర్యను అమలు చేయడానికి మోడల్ స్పందనను చదవడం అంటే ఫంక్షన్ లేదా API కాల్ ను నడపడం 
3. ఫంక్షన్ నుండి వచ్చిన_RESP_తో చాట్ కంప్లీషన్స్ APIని మరోసారి కాల్ చేయడం, ఆ సమాచారాన్ని ఉపయోగించి వినియోగదారుకు స్పందనను సృష్టించడం. 


![ఫంక్షన్ కాల్ యొక్క ప్రవాహం](../../../../translated_images/te/LLM-Flow.3285ed8caf4796d7.webp)


### ఫంక్షన్ కాల్ యొక్క మూలకాలు

#### ఉపయోగకర్త ఇన్‌పుట్

మొదటి దశ ఒక వినియోగదారు సందేశాన్ని సృష్టించడం. ఇది టెక్స్ట్ ఇన్‌పుట్ విలువ తీసుకోవడం ద్వారా గమనిస్తుండవచ్చు లేక మీరు ఇక్కడ విలువను కేటాయించవచ్చు. ఇది మీ మొదటి సారిగా చాట్ కంప్లీషన్స్ API తో పని చేస్తుంటే, మేము సందేశం యొక్క `role` మరియు `content` ను నిర్వచించాలి.

`role` ఏమన్నా `system` (నియమాలు సృష్టించటం), `assistant` (మోడల్) లేదా `user` (అంతిమ వినియోగదారు) కావచ్చు. ఫంక్షన్ కాలింగ్ కోసం, మేము దీన్ని `user` గా కేటాయించి ఒక ఉదాహరణ ప్రశ్న ఇవ్వబోతున్నాము.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ఫంక్షన్లు సృష్టించడం. 

ఆ తర్వాత మేము ఒక ఫంక్షన్ మరియు ఆ ఫంక్షన్ యొక్క పరామితులను నిర్వచిస్తాము. ఇక్కడ మేము కేవలం ఒక ఫంక్షన్ `search_courses` ఉపయోగిస్తాము కానీ మీరు అనేక ఫంక్షన్లు సృష్టించవచ్చు.

**మहत్త్వం** : ఫంక్షన్లు LLM కు సిస్టమ్ సందేశంలో చేర్చబడతాయి మరియు మీరు ఉపయోగించదగిన టోకన్ల యెక్క పరిమితిలో ఉన్నాయి. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

** నిర్వచనాలు ** 

ఫంక్షన్ నిర్వచన నిర్మాణం అనేక స్థాయిలును కలిగి ఉంది, ప్రతి స్థాయికి దాని సొంత లక్షణాలు ఉంటాయి. ఇక్కడ నెస్టెడ్ నిర్మాణం ఒక విభజన ఉంది:

** టాప్ లెవెల్ ఫంక్షన్ లక్షణాలు:**

`name` - మేము పిలవదలచుకున్న ఫంక్షన్ పేరు.

`description` - ఇది ఫంక్షన్ ఎలా పని చేస్తుందో వివరణ. ఇక్కడ స్పష్టంగా మరియు నిర్దిష్టంగా ఉండటం ముఖ్యం

`parameters` - మీరు మోడల్ దాని ప్రతిస్పందనలో ఉత్పత్తి చేయించాలని కోరుకునే విలువలు మరియు ఫార్మాట్ యొక్క జాబితా

** ప్యారామీటర్లు ఆబ్జెక్ట్ లక్షణాలు:**

`type` - ప్యారామీటర్ల ఆబ్జెక్ట్ యొక్క డేటా రకం (సాధారణంగా "object")

`properties` - మోడల్ తన ప్రతిస్పందన కోసం ఉపయోగించే ప్రత్యేక విలువల జాబితా

** వ్యక్తిగత ప్యారామీటర్ లక్షణాలు:**

`name` - ప్రాపర్టీ కీ ద్వారా నిమిషంగా నిర్వచించబడింది (ఉదా: "role", "product", "level")

`type` - ఈ ప్రత్యేక ప్యారామీటర్ యొక్క డేటా రకం (ఉదా: "string", "number", "boolean")

`description` - ఈ ప్రత్యేక ప్యారామీటర్ యొక్క వివరణ

** ఐచ్ఛిక లక్షణాలు:**

`required` - ఫంక్షన్ కాల్ పూర్తిచేయడానికి ఏ ప్యారామీటర్లు అవసరం అనేది చూపించే ఒక అర్రే


### ఫంక్షన్ కాల్ చేయడం  
ఒక ఫంక్షన్ నిర్వచించిన తర్వాత, ఇప్పుడు దాని Chat Completion APIకు కాల్‌లో చేర్చాలి. మేము ఇది `functions`ను రిక్వెస్ట్‌కు జోడించడం ద్వారా చేస్తాము. ఈ సందర్భంలో `functions=functions`.  

`function_call`ను `auto`గా సెట్ చేయాలనే ఎంపిక కూడా ఉంది. అంటే, వినియోగదారుని సందేశానుసారం ఏ ఫంక్షన్ కాల్ చేయాలో LLM నిర్ణయించనీయబడుతుంది, మేము అది స్వయంగా నిర్ణయించుకోము.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ఇప్పుడు ప్రతిస్పందనను చూద్దాం మరియు అది ఎలా ఫార్మాట్ చేయబడిందో చూడండి: 

{
  "role": "assistant",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

మీరు గమనించవచ్చు ఫంక్షన్ పేరు పిలవబడింది మరియు యూజర్ సందేశం నుండి, LLM ఆర్గుమెంట్స్ ను సరిపోతున్నట్టుగా డేటాను కనుగొనగలిగింది. 


## 3. అప్లికేషన్‌లో ఫంక్షన్ కాల్స్‌ను అనుసంధానించడం.


LLM నుండి రూపొంది ఉన్న ప్రతిస్పందనను పరీక్షించిన తర్వాత, ఇప్పుడు ఈ తదుపరి దాన్ని ఒక అప్లికేషన్‌లో అనుసంధానించవచ్చు.

### ప్రవాహాన్ని నిర్వహించడం

దీన్ని మా అప్లికేషన్‌లో అనుసంధానించడానికి, కింది చర్యలను తీసుకుందాము:

మొదటగా, OpenAI సేవలకు కాల్ చేయండి మరియు సందేశాన్ని `response_message` అనే వేరియబుల్‌లో నిల్వ చేయండి.


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ఇప్పుడు మేము కోర్సుల జాబితా పొందడానికి Microsoft Learn API ను పిలిచే ఫంక్షన్ ను నిర్వచించబోతున్నాము: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ఉత్తమ ఆచరణగా, మodel్ ఫంక్షన్‌ను పిలుచుకోవాలనుకుంటుందా అని మేము చూస్తాము. ఆ తర్వాత, అందుబాటు లో ఉన్న ఫంక్షన్లలో ఒకటిని సృష్టించి, పిలవబడుతున్న ఫంక్షన్‌కు అనుసంధానం చేస్తాము.
తర్వాత, ఆ ఫంక్షన్ యొక్క ఆర్గ్యూమెంట్లను తీసుకొని, అవి LLM నుండి వచ్చిన ఆర్గ్యూమెంట్లకు మ్యాప్ చేస్తాము.

చివరగా, ఫంక్షన్ కాల్ సందేశం మరియు `search_courses` సందేశం ద్వారా తిరిగి వచ్చిన విలువలను జతచేస్తాము. ఇది LLM కి సహజ భాషలో వినియోగదారుని జవాబు ఇవ్వడానికి అవసరమైన అన్ని సమాచారాన్ని ఇస్తుంది.
వినియోగదారుడికి సహజ భాషలో స్పందించడానికి.

ఇప్పుడు మేము నవీకరించబడిన సందేశాన్ని LLMకు పంపబోతున్నాము, తద్వారా మేము API JSON ఆకృతుల లోకి కాకుండా సహజ భాష ప్రతిస్పందనను అందుకోవచ్చు.
