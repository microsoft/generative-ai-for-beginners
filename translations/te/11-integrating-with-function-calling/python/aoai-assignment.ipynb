## ఫంక్షన్ కాల్స్ ని అర్థం చేసుకోవడం  

ఈ పాఠం కోసం, మేము మా విద్యా స్టార్టప్‌కి ఒక ఫీచర్‌ను తయారు చేయాలనుకుంటున్నాము, ఇది వినియోగదారులకు టెక్నికల్ కోర్సులను కనుగొనడానికి చాట్‌బాట్‌ను ఉపయోగించడానికి అనుమతిస్తుంది. మేము వారి నైపుణ్య స్థాయి, ప్రస్తుత పాత్ర మరియు ఆసక్తి ఉన్న సాంకేతికత మరియు సరిపోయే కోర్సులను సిఫారసు చేస్తాము.  

దీన్ని పూర్తి చేయడానికి మేము ఈ కాంబినేషన్‌ని ఉపయోగిస్తాము:  
 - వినియోగదారులకు చాట్ అనుభవాన్ని సృష్టించడానికి `Azure Open AI`  
 - వినియోగదారుల అభ్యర్థన ఆధారంగా కోర్సులు కనుగొనడానికి సహాయం చేయడానికి `Microsoft Learn Catalog API`  
 - వినియోగదారుల ప్రశ్నను తీసుకుని API అభ్యర్థన చేయడానికి ఫంక్షన్‌కు పంపేందుకు `Function Calling`  

ప్రారంభించడానికి, మేము ఎందుకు ఫంక్షన్ కాలింగ్ ఉపయోగించాలనుకుంటున్నామో చూద్దాం:  


### ఫంక్షన్ కాలింగ్ ఎందుకు 

మీరు ఈ కోర్సులోని ఇతర పాఠ్య భాగాలను పూర్తి చేసి ఉంటే, మీరు పెద్ద భాషా నమూనాలను (LLMs) ఉపయోగించే శక్తిని అర్థం చేసుకున్నట్లుంటారు. దానితో పాటు, వాటి కొన్ని పరిమితులను కూడా మీరు చూడగలుగుతారు. 

ఫంక్షన్ కాలింగ్ అనేది ఆజ్యూర్ ఓపెన్ AI సర్వీస్ యొక్క ఒక ఫీచర్, దీని ద్వారా క్రింది పరిమితులను అధిగమించవచ్చు: 
1) స్థిరమైన రిస్పాన్స్ ఫార్మాట్ 
2) చాట్ సందర్భంలో అప్లికేషన్ యొక్క ఇతర మూలాల నుండి డేటాను ఉపయోగించే సామర్థ్యం 

ఫంక్షన్ కాలింగ్ ముందు, LLM నుండి వచ్చే సమాధానాలు అసంఘటితం మరియు స్థిరచేయని ఉండేవి.  డెవలపర్లు ప్రతి వేరియేషన్‌ను హ్యాండిల్ చేయడానికి సంక్లిష్టమైన ధృవీకరణ కోడ్ రాయాల్సిన అవసరం ఉండేది. 

వినియోగదారులు "స్టాక్‌హోంన్లో ప్రస్తుత వాతావరణం ఏమిటి?" వంటి జవాబులు పొందలేకపోయే అవకాశముంది. ఎందుకంటే నమూనాలు శిక్షణ పొందిన డేటా సమయం వరకు పరిమితంగా ఉండేవి. 

ఈ సమస్యను వివరించే దిగువ ఉదాహరణను చూద్దాం: 

మనం ఒక విద్యార్థి డేటాబేస్ క్రియేట్ చేస్తామని తీసుకుందాం, తద్వారా వారికి సరైన కోర్సును సూచించవచ్చు. దిగువ రెండు విద్యార్థుల వివరణలు ఉన్నాయి, ఇవి ఉన్న డేటాలో చాలా సమానమైనవి. 


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ఈ డేటాను విశ్లేషించడానికి మేము దీన్ని LLM కు పంపించాలని కోరుకుంటున్నాము. ఇది తరువాత మా యాప్‌లో API కి పంపడానికి లేదా డేటాబేస్‌లో నిల్వ చేయడానికి ఉపయోగించవచ్చు. 

మేము LLM కు ఆసక్తి ఉన్న సమాచారాన్ని సూచించే రెండు ఒకటే ప్రాంప్ట్‌లను సృష్టిద్దాం: 


మన ఉత్పత్తికి ముఖ్యమైన భాగాలను విశ్లేషించడానికి దీన్ని ఒక LLM కు పంపాలని మనము కోరుతున్నాము. కాబట్టి మనము LLM కు సూత్రీకరించడానికి రెండు సమానమైన సూచనలు సృష్టించగలగాలి: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ఈ రెండు ప్రాంప్టులను సృష్టించిన తర్వాత, మేము వాటిని `client.responses.create` ఉపయోగించి LLM కు పంపుతాము. మేము ప్రాంప్ట్‌ను `input` అనే వేరియబుల్‌లో నిల్వ చేస్తాము మరియు పాత్రను `user` కు割��హిస్తాము. ఇది ఒక యూజర్ chatbot కు ఒక సందేశం పంపుతున్నట్లుగా ఉద్దేశించడానికి.  



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ఇప్పడు మనం ఎల్ఎండికి ఈ రెండు అభ్యర్థనలను పంపి, మనం పొందే స్పందనను పరిశీలించవచ్చు. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ప్రాంప్ట్స్ ఒకే విధంగా ఉండి, వివరణలు సాదృశ్యంగా ఉన్నా, మేము `Grades` లక్షణానికి వేర్వేరు ఫార్మాట్లు పొందవచ్చు.

పై సెల్‌ని పలు సార్లు నడపితే, ఫార్మాట్ `3.7` లేదా `3.7 GPA` కావచ్చు.

ఇది ఎందుకంటే LLM అంసరహిత డేటా లిఖిత ప్రాంప్ట్ ఆకారంలో తీసుకుని, అంసరహిత డేటాను తిరిగి ఇస్తుంది. మనం ఒక నిర్మిత ఫార్మాట్ అవసరం, తద్వారా ఈ డేటాను నిల్వ చేయడం లేదా ఉపయోగించడం సమయం మనం ఏమి ఆశించాలో తెలుసుకోవచ్చు.

ఫంక్షనల్ కాలింగ్ ఉపయోగించి, మనం నిర్మిత డేటాను తిరిగి పొందడాన్ని నిర్ధారించుకోవచ్చు. ఫంక్షన్ కాలింగ్ ఉపయోగించినప్పుడు, LLM నిజానికి ఎలాంటి ఫంక్షన్లను పిలవదు లేదా నడిపించదు. బదులుగా, LLM స్పందనల కోసం అనుసరించవలసిన నిర్మాణాన్ని మనం సృష్టిస్తాము. ఆ తరువాత ఆ నిర్మిత స్పందనలను మన అప్లికేషన్లలో ఏ ఫంక్షన్ నడపాలో తెలుసుకోవడానికి ఉపయోగిస్తాము.

### ఫంక్షన్ కాల్స్ ఉపయోగించే ఉపయోగాల సందర్భాలు 

**బాహ్య టూల్స్ కాలింగ్**
చాట్బాట్లు వినియోగదారుల నుండి వచ్చిన ప్రశ్నలకు సమాధానాలు అందించడం గొప్పగా చేస్తాయి. ఫంక్షన్ కాలింగ్ ఉపయోగించడం ద్వారా, చాట్బాట్లు వినియోగదారుల సందేశాలను ఉపయోగించి కొన్ని పనులను పూర్తి చేయవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి చాట్బాట్‌కి "ఈ సబ్జెక్ట్‌లో నాకు మరింత సహాయం కావాలి అని నా ఇన్స్ట్రక్టర్‌కి ఈమెయిల్ పంపించు" అని అడగవచ్చు. ఇది `send_email(to: string, body: string)` అనే ఫంక్షన్ కాల్ చేయవచ్చు


**API లేదా డేటాబేస్ క్వెరీలు సృష్టించడం**
వినియోగదారులు సహజ భాష ఉపయోగించి సమాచారాన్ని కనుగొనవచ్చు, ఇది ఫార్మాటెడ్ క్వెరీ లేదా API అభ్యర్థనగా మారుతుంది. ఉదాహరణకి, ఒక ఉపాధ్యాయుడు "ముందరి అసైన్మెంట్‌ను పూర్తి చేసిన విద్యార్థులు ఎవరు" అని అడగడంతో `get_completed(student_name: string, assignment: int, current_status: string)` అనే ఫంక్షన్‌కి కాల్ చేయవచ్చు


**సంరచిత డేటా సృష్టించడం**
వినియోగదారులు ఒక టెక్ట్స్ లేదా CSV బ్లాక్ తీసుకుని LLM ఉపయోగించి ముఖ్యమైన సమాచారాన్ని సేకరించవచ్చు. ఉదాహరణకి, ఒక విద్యార్థి శాంతి ఒప్పందాల గురించి వికీపీడియా ఆర్టికల్‌ను AI ఫ్లాష్ కార్డ్స్‌గా మార్చవచ్చు. ఇది `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` అనే ఫంక్షన్ ఉపయోగించి చేయవచ్చు


## 2. మీ మొదటి ఫంక్షన్ కాల్ సృష్టించడం  

ఫంక్షన్ కాల్ సృష్టించే ప్రక్రియలో 3 ప్రధాన దశలు ఉంటాయి:  
1. మీ ఫంక్షన్ల జాబితా మరియు వినియోగదారుని సందేశంతో చాట్ కంప్లీషన్స్ API ను కాల్ చేయడం  
2. మోడల్ ప్రతిస్పందనను చదివి చర్య తీసుకోవడం అంటే ఫంక్షన్ లేదా API కాల్‌ను అమలు చేయడం  
3. వినియోగదారునికి ప్రక్రియను సృష్టించే ఆ సమాచారాన్ని ఉపయోగించడానికి మీ ఫంక్షన్ నుండి వచ్చిన ప్రతిస్పందనతో మరోసారి చాట్ కంప్లీషన్స్ API ను కాల్ చేయడం

### ఫంక్షన్ కాల్ యొక్క అంశాలు

#### వినియోగదారుల ఇన్‌పుట్

మొదటి దశ ఒక వినియోగదారు సందేశాన్ని సృష్టించడం. ఇది ఒక టెక్స్ట్ ఇన్‌పుట్ విలువను పొందడం ద్వారా డైనమిక్‌గా కేటాయించవచ్చు లేదా మీరు ఇక్కడ విలువను కేటాయించవచ్చు. ఇది మీరు Chat Completions APIతో మొదటిసారిగా పని చేస్తున్నట్లయితే, మేము సందేశపు `role` మరియు `content` ను నిర్వచించాలి.

`role` అంటే `system` (నియమాలు సృష్టించడం), `assistant` (మోడల్) లేదా `user` (చివరి వినియోగదారు) ఉండవచ్చు. ఫంక్షన్ కాలింగ్ కోసం, మేము దీనిని `user` గా కేటాయించి ఒక ఉదాహరణ ప్రశ్నను ఇస్తాము.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ఫంక్షన్లు సృష్టించడం. 

తరువాత మనం ఒక ఫంక్షన్ మరియు ఆ ఫంక్షన్ యొక్క పరామితులను నిర్వచిస్తాము. ఇక్కడ మనం కేవలం ఒక ఫంక్షన్ `search_courses` అనే పేరు తో మాత్రమే ఉపయోగించబోతున్నాము కానీ మీరు అనేక ఫంక్షన్లు సృష్టించవచ్చు.

**ముఖ్యమైనది** : ఫంక్షన్లు LLMకు సిస్టమ్ సందేశంలో చేర్చబడతాయి మరియు ఇవి మీరు ఉపయోగించగల టోకెన్ల పరిమాణంలో కూడా చేర్చబడతాయి. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**定義లు** 

`name` - మేము పిలవాలని కోరుకునే ఫంక్షన్ పేరు. 

`description` - ఫంక్షన్ ఎలా పనిచేస్తుంది అనే వివరణ. ఇక్కడ స్పష్టంగా మరియు సూటిగా ఉండటం ముఖ్యము. 

`parameters` - మీరు మోడల్ స్పందనలో ఉత్పత్తి చేయాలని కోరుకునే విలువల మరియు ఫార్మాట్ జాబితా. 


`type` - లక్షణాల డేటా రకం నిల్వ చేయబడుతుంది. 

`properties` - మోడల్ స్పందన కోసం ఉపయోగించే నిర్దిష్ట విలువల జాబితా. 


`name` - మోడల్ తన ఫార్మాటెడ్ స్పందనలో ఉపయోగించే గుణధర్మం యొక్క పేరు. 

`type` - ఈ గుణధర్మం యొక్క డేటా రకం. 

`description` - నిర్దిష్ట గుణధర్మం యొక్క వివరణ. 


**ఐచ్ఛికం**

`required` - ఫంక్షన్ కాల్ పూర్తి కావడానికి తప్పనిసరి గుణధర్మం. 


### ఫంక్షన్ కాల్ చేయడం  
ఒక ఫంక్షన్‌ను నిర్వచించిన తర్వాత, మేము దీనిని Chat Completion API కాల్‌లో చేర్చాలి. దీని కోసం, మేము అభ్యర్థనలో `functions`‌ను జోడిస్తాము. ఈ సందర్భంలో `functions=functions`.  

`function_call`ను `auto`గా సెట్ చేయడం కోసం కూడా ఒక ఎంపిక ఉంది. దీని అర్థం, వినియోగదారుడి సందేశం ఆధారంగా ఎలాంటి ఫంక్షన్ కాల్ చేయాలో LLM నిర్ణయిస్తుంది, మేము స్వయంగా నియమించుకోవడం కాదని అని.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ఇప్పుడింతానే ప్రతిస్పందనని చూద్దాం మరియు అది ఎలా ఫార్మాట్ చేయబడిందో గమనిస్తాం: 

{
  "role": "సహాయకుడు",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

మీరు గమనించవచ్చు, ఫంక్షన్ పేరును కాల్ చేశారు మరియు యూజర్ సందేశం నుండి LLM ఆ ఫంక్షన్ యొక్క ఆర్గ్యుమెంట్లకు సరిపోయే డేటాను కనుగొన్నారు. 


## 3.ఫంక్షన్ కాల్స్‌ను అనువర్తనంలో సమగ్రపరచడం. 


LLM నుండి ఫార్మాట్ చేసిన ప్రతిస్పందనను పరీక్షించిన తరువాత, ఇప్పుడు దీన్ని ఒక అనువర్తనంలో సమగ్రపరచవచ్చు. 

### ప్రవాహం నిర్వహణ 

దీన్ని మా అనువర్తనంలో సమగ్రపరచడానికి, క్రింది దశలను తీసుకుందాం: 

మొదట, Open AI సేవలకి కాల్ చేసి సందేశాన్ని `response_message` అనే వెరియబుల్‌లో నిల్వచేయుదాం. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ఇప్పుడు మేము Microsoft Learn API ని పిలిపించి కోర్సుల జాబితాను పొందడానికి ఉపయోగించే ఫังก్షన్‌ని నిర్వచించబోతున్నాము: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ఉత్తమ ఆచారంగా, మోడల్ ఒక ఫంక్షన్ ని కాల్ చేయాలని కోరుకుంటుందా అని చూసుకోడమే. ఆ తర్వాత, అందుబాటులో ఉన్న ఫంక్షన్ లలో ఒకటిని సృష్టించి, కాల్ అవుతున్న ఫంక్షన్ కు సరిపోల్చతాము.
తర్వాత, ఫంక్షన్ యొక్క ఆర్గ్యూమెంట్లను LLM నుండి వచ్చిన ఆర్గ్యూమెంట్లకు మ్యాప్ చేస్తాము.

చివరగా, ఫంక్షన్ కాల్ సందేశం మరియు `search_courses` సందేశం ద్వారా తిరిగి వచ్చిన విలువలను జోడిస్తాము. ఇది LLM కి
సహజ భాషలో వినియోగదారుడికి స్పందించడానికి అవసరమైన మొత్తం సమాచారం అందిస్తుంది.
