## ఫంక్షన్ కాల్స్ అర్థం చేసుకోవడం 

ఈ పాఠం కోసం, మన విద్య ఉత్పత్తి ప్రారంభ సంస్థకు ఒక ఫీచర్ సృష్టించాలనుకుంటున్నాము, ఇది వినియోగదారులు టెక్నికల్ కోర్సులను కనుగొనటానికి చాట్‌బాట్ ఉపయోగించడానికి అనుమతిస్తుంది. వారి నైపుణ్య స్థాయి, ప్రస్తుత పాత్ర మరియు ఆసక్తి ఉన్న సాంకేతికతకు అనుగుణంగా కోర్సులను మేము సిఫారసు చేస్తాము. 

దీన్ని పూర్తి చేయడానికి మేము క్రింది కలయికను ఉపయోగిస్తాము: 
 - `Azure Open AI` వినియోగదారులకు చాట్ అనుభవం సృష్టించడానికి
 - `Microsoft Learn Catalog API` వినియోగదారుల అభ్యర్థన ఆధారంగా కోర్సులను కనుగొనడంలో సహాయపడడానికి 
 - `Function Calling` వినియోగదారుల ప్రశ్నను తీసుకుని, API అభ్యర్థన చేయడానికి ఒక ఫంక్షన్‌కు పంపడానికి. 

ప్రారంభించడానికి, మనం మొదట ఎందుకు ఫంక్షన్ కాలింగ్ ఉపయోగించాలి అనేది చూద్దాం: 


### ఫంక్షన్ కాల్ చేయడానికి కారణం  

మీరు ఈ కోర్సులో మరేదైనా పాఠం పూర్తి చేసి ఉంటే, మీరు పెద్ద భాషా మోడల్స్ (LLMs) ఉపయోగించే శక్తిని అర్థం చేసుకున్నట్లు ఉంటారు. ఆశంగా మీరు వాటి కొన్ని పరిమితులను కూడా చూడగలుగుతారు.  

ఫంక్షన్ కాలింగ్ అనేది Azure Open AI సర్వీస్ యొక్క ఒక ఫీచర్, దిగువ పరిమితులను అధిగమించడానికి:  
1) సాదృశ్యమైన స్పందన ఫార్మాట్  
2) చాట్ సందర్భంలో అనువర్తనంలోని ఇతర వనరుల డేటాను ఉపయోగించే సామర్థ్యం  

ఫంక్షన్ కాలింగ్ ముందు, LLM నుండి వచ్చిన స్పందనలు అసంగతమైనవిగా మరియు సాదృశ్యంలేనివిగా ఉండేవి. వృత్తి నిపుణులు ప్రతి ఒక్క రకాల స్పందనను సరిగా నిర్వహించడానికి క్లిష్టమైన ధృవీకరణ కోడ్ రాయాల్సి ఉండింది.  

వినియోగదారులు "స్టాక్‌హోల్‌మ్‌లో కర్తమ కాలంలో వాతావరణం ఎలా ఉంది?" లాంటి ప్రశ్నలకు సమాధానం పొందలేకపోయారు. ఇది కారణం మోడల్స్ శిక్షణ పొందిన డేటా సమయానికి పరిమితమయ్యాయి.  

ఈ సమస్యను వివరిస్తున్న ఉదాహరణను దిగువ చూడండి:  

మనం విద్యార్థుల డేటా డేటాబేస్‌ను సృష్టించి, వారికి సరైన కోర్సును సూచించే ఉద్దేశ్యంతో ఉన్నామనుకుందాం. దిగువ న two ఇద్దరు విద్యార్థుల వివరణలు ఉన్నాయి వాటి డేటాలో చాలా సమానం.  


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


మేము ఈ డేటాను పార్స్ చేయడానికి ఒక LLMకి పంపాలని కోరుకుంటున్నాము. ఇది తరువాత మా అప్లికేషన్‌లో APIకి పంపడానికి లేదా డేటాబేస్‌లో నిల్వ చేయడానికి ఉపయోగించవచ్చు.

మేము రెండు సమానమైన ప్రాంప్ట్‌లను సృష్టిద్దాం, వాటిలో LLMకి మేము ఆసక్తి ఉన్న సమాచారాన్ని సూచిస్తాము:


మా ఉత్పత్తికి ముఖ్యమైన భాగాలను విభజించడానికి దీన్ని ఒక LLMకు పంపించాలని మనం కోరుకుంటున్నాము. కాబట్టి LLMకు సూచించడానికి రెండు సరిగ్గా ఒకే_PROMPTS_ను సృష్టించవచ్చు: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ఈ రెండు ప్రాంప్ట్‌లను సృష్టించిన తర్వాత, మేము వాటిని LLM కు `client.responses.create` ఉపయోగించి పంపిస్తాము. మేము ప్రాంప్ట్‌ను `input` వేరియబుల్‌లో నిల్వచేస్తాము మరియు పాత్రను `user` కు కేటాయిస్తాము. ఇది యూజర్ నుండి చాట్‌బాట్‌కి ఒక సందేశం రాశామనేదాన్ని అనుకరించడానికి. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ఇప్పుడు మనం రెండు అభ్యర్థనలను LLMకి పంపి మనం అందుకున్న ప్రతిస్పందనను పరిశీలించవచ్చు.


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


ప్రాంప్టులు ఒకటే అయినా, వివరణలు సమానమైనా, `Grades` ప్రాపర్టీ యొక్క భిన్న ఫార్మాట్లను పొందవచ్చు. 

పై సెల్‌ను పలు సార్లు నడపితే, ఫార్మాట్ `3.7` లేదా `3.7 GPA` కావచ్చు. 

దీని కారణం LLM అనేది రాయించిన ప్రాంప్ట్ రూపంలో నిర్మాణ రహిత డేటాను తీసుకుని, నిర్మాణ రహిత డేటాను తిరిగి పంపడం. డేటా నిల్వ చేయడం లేదా ఉపయోగించేటప్పుడు ఏమి ఆశించాలో మనకు తెలుసుకోవడానికి నిర్మాణాత్మక ఫార్మెట్ అవసరం. 

ఫంక్షనల్ కాలింగ్ ఉపయోగించడం ద్వారా మనం నిర్మాణాత్మక డేటాను తిరిగి పొందగలుగుతాము. ఫంక్షన్ కాలింగ్ ఉపయోగించినప్పుడు, LLM వాస్తవానికి ఏ ఫంక్షన్లను కాల్ చేయదు లేదా నడపదు. దాని ప్రత్యామ్నాయంగా, మనం LLM అనుసరించాల్సిన నిర్మాణాన్ని సృష్టిస్తాము. ఆ నిర్మాణాత్మక ప్రతిస్పందనలను ఉపయోగించి మన అప్లికేషన్లలో ఏ ఫంక్షన్ నడపాలో తెలుసుకుంటాము.

### ఫంక్షన్ కాల్‌లను ఉపయోగించే వినియోగ కేసులు 

**బయటి సాధనాలను కాల్ చేయడం** 
చాట్బాట్‌లు వాడుకరుల ప్రశ్నలకు సమాధానాలు ఇవ్వడంలో గొప్పవిగా ఉంటాయి. ఫంక్షన్ కాలింగ్ ఉపయోగించడం ద్వారా, చాట్బాట్‌లు వాడుకరుల మెసేజెస్‌ను ఉపయోగించి కొన్ని పనులు పూర్తి చేయవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి “ఈ విషయం గురించి నాకు మరింత సహాయం అవసరం అని నా ఇన్‌స్ట్రక్టర్‌కు ఇమెయిల్ పంపించు” అని చాట్‌బాట్‌కు అడగవచ్చు. ఇది `send_email(to: string, body: string)` అనే ఫంక్షన్ కాల్ చేయవచ్చు


**API లేదా డేటాబేస్ క్వెరీలను సృష్టించడం**
వాడుకర్లు సహజ భాషను ఉపయోగించి సమాచారాన్ని కనుగొనవచ్చు, అది ఫార్మatted క్వెరీ లేదా API అభ్యర్థనగా మారుతుంది. దీని ఉదాహరణగా ఒక ఉపాధ్యాయుడు “కొత్తగా పూర్తి చేసిన అసైన్‌మెంట్ పూర్తి చేసిన విద్యార్థులు ఎవరు?” అని అడుగుతాడు, ఇది `get_completed(student_name: string, assignment: int, current_status: string)` అనే ఫంక్షన్‌ను కాల్ చేయవచ్చు


**రూపుకల్పితం డేటాను సృష్టించడం**
వాడుకర్లు ఒక బ్లాక్ టెక్స్ట్ లేదా CSV తీసుకొని, దానిలో నుంచి ముఖ్యమైన సమాచారం పొందడానికి LLM ఉపయోగించవచ్చు. ఉదాహరణకు, ఒక విద్యార్థి శాంతి ఒప్పందాల గురించి వికీపీడియా ఆర్టికల్‌ను AI ఫ్లాష్ కార్డ్లు సృష్టించేందుకు మార్చవచ్చు. ఇది `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` అనే ఫంక్షన్ ఉపయోగించి చేయవచ్చు


## 2. మీ మొదటి ఫంక్షన్ కాల్‌ను సృష్టించడం 

ఫంక్షన్ కాల్‌ను సృష్టించే ప్రక్రియలో 3 ముఖ్యమైన దశలు ఉంటాయి: 
1. మీ ఫంక్షన్లు మరియు ఒక యూజర్ సందేశం జాబితాతో చాట్ కంప్లీషన్స్ APIను కాల్ చేయడం 
2. పని నిర్వహించేందుకు మోడల్ స్పందనను చదవడం, అంటే ఒక ఫంక్షన్ లేదా API కాల్‌ను అమలు చేయడం 
3. యూజర్‌కు ప్రతిస్పందన సృష్టించడానికి ఆ సమాచారాన్ని ఉపయోగించుకుని మీ ఫంక్షన్ నుండి వచ్చిన ప్రతిస్పందనతో చాట్ కంప్లీషన్స్ APIను మళ్ళీ కాల్ చేయడం.

### ఫంక్షన్ కాల్ యొక్క మూలకాలు 

#### ఉపయోగదారుల ఇన్‌పుట్ 

మొదటి దశ ఒక వినియోగదారు సందేశాన్ని సృష్టించడమే. ఇది ఒక టెక్స్ట్ ఇన్‌పుట్ విలువను తీసుకొని డైనమిక్‌గా కేటాయించవచ్చు లేదా మీరు ఇక్కడ ఒక విలువను కేటాయించవచ్చు. ఇది Chat Completions APIతో మీరు మొదటి రౌండ్ అయితే, మేము సందేశం యొక్క `role` మరియు `content` ను నిర్వచించాల్సి ఉంటుంది. 

`role` కావాలంటే `system` (నియమాలను సృష్టించడం), `assistant` (మోడల్) లేదా `user` (ఇండ-యూజర్) కావచ్చు. ఫంక్షన్ కాలింగ్ కోసం, మేము దీన్ని `user` గా కేటాయించి ఒక ఉదాహరణ ప్రశ్నను ఇస్తాము. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ఫంక్షన్లు సృష్టించడం. 

తదుపరి మేము ఒక ఫంక్షన్ మరియు ఆ ఫంక్షన్ యొక్క పరిమితులను నిర్వచిస్తాము. ఇక్కడ `search_courses` అని ఒకే ఒక ఫంక్షన్ ఉపయోగిస్తాము కాని మీరు అనేక ఫంక్షన్లను సృష్టించుకోవచ్చు.

**ముఖ్యం** : ఫంక్షన్లు LLM కి సిస్టమ్ సందేశంలో చేర్చబడ్డాయి మరియు మీరు అందుబాటులో ఉన్న టోకెన్ల పరిమితిలో భాగంగా ఉంటాయి. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**నిర్వచనలు** 

`name` - మేము పిలిచించదలచిన ఫంక్షన్ పేరు. 

`description` - ఫంక్షన్ ఎలా పనిచేస్తుందో వివరణ. ఇక్కడ స్పష్టంగా మరియు ఖచ్చితంగా ఉండటం ముఖ్యం 

`parameters` - మోడల్ రిప్లైలో ఉత్పత్తి చేయాలనుకునే విలువల జాబితా మరియు ఫార్మాట్ 


`type` - ప్రాపర్టీల డేటా రకం నిల్వ చేయబడుతుంది 

`properties` - రిప్లై కోసం మోడల్ ఉపయోగించే నిర్దిష్ట విలువల జాబితా 


`name` - ఫార్మాటెడ్ రిప్లైలో మోడల్ ఉపయోగించే ప్రాపర్టీ పేరు 

`type` - ఈ ప్రాపర్టీకి సంబంధించిన డేటా రకం 

`description` - నిర్దిష్ట ప్రాపర్టీ వివరించినది 


**ఐచ్ఛికం**

`required` - ఫంక్షన్ కాల్ పూర్తి చేయడానికి అవసరమైన ప్రాపర్టీ 


### ఫంక్షన్ కాల్ చేయడం  
ఒక ఫంక్షన్‌ను నిర్వచించిన తర్వాత, ఇప్పుడు దాన్ని Chat Completion API కాల్‌లో చేర్చాల్సి ఉంటుంది. మేము దీనిని అభ్యర్థనలో `functions` ను చేర్చడం ద్వారా చేస్తాము. ఈ సందర్భంలో `functions=functions`.  

`function_call` ని `auto` గా సెట్ చేసే ఒక ఎంపిక కూడా ఉంది. దీనర్థం ఏమిటంటే, యూజర్ సందేశం ఆధారంగా ఏ ఫంక్షన్‌ను కాల్ చేయాలో LLM నిర్ణయిస్తుంది, మేమే కేటాయించవద్దు.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

ఇప్పుడు మనం ప్రతిస్పందనను చూసి ఇది ఎలా రూపకల్పన చేయబడిందో చూద్దాం: 

{
  "role": "assistant",
  "function_call": {
    "name": "search_courses",
    "arguments": "{\n  \"role\": \"student\",\n  \"product\": \"Azure\",\n  \"level\": \"beginner\"\n}"
  }
}

మీరు చూడగలరు ఈ ఫంక్షన్కు పేరు పిలవబడింది మరియు వాడుకరి సందేశం నుండి, LLM ఫంక్షన్ ఆర్కుమెంట్స్ కు సరిపోయే డేటాను కనుగొన్నారు. 


## 3.ఒక అనువర్తనంలో ఫంక్షన్ కాల్స్‌ను ఒకేసారి చేర్చడం. 


మనం LLM నుండి ఫార్మాటెడ్ స్పందనను పరీక్షించిన తర్వాత, దీన్ని ఒక అనువర్తనంలో చేర్చవచ్చు. 

### ప్రవాహాన్ని నిర్వహించడం 

దీన్ని మన అనువర్తనంలో చేర్చడానికి, క్రింది దశలను తీసుకుందాం: 

మొదటిగా, Open AI సేవలకు కాల్ చేసి, `response_message` అని పేరున్న వెరీయబుల్‌లో సందేశాన్ని నిల్వచేయండి. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ఇప్పుడు మేము కోర్సుల జాబితాను పొందడానికి Microsoft Learn API ని పిలుచుకునే ఫంక్షన్‌ను నిర్వచించబోతున్నాము: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ఉత్తమ ఆచారంగా, మేము తర్వాత మోడల్ ఫంక్షన్‌ను పిలవాలనుకుంటుందా అని చూద్దాము. ఆ తర్వాత, అందుబాటులో ఉన్న ఫంక్షన్లలో ఒకదాన్ని సృష్టించి పిలవబడిస్తున్న ఫంక్షన్‌కు సరిపోండి. 
మేము ఆ తర్వాత ఫంక్షన్ యొక్క ఆర్గ్యుమెంట్లను తీసుకుని అవి LLM నుండి పొందిన ఆర్గ్యుమెంట్లకు మ్యాప్ చేస్తాము.

చివరగా, మేము ఫంక్షన్ కాల్ సందేశం మరియు `search_courses` సందేశం ద్వారా తిరిగి వచ్చిన విలువలను జోడిస్తాము. ఇది LLM కి 
సహజ భాషను ఉపయోగించి వినియోగదారుడికి స్పందించడానికి కావలసిన మొత్తం సమాచారాన్ని అందిస్తుంది.
