## ల్లామా 3.1 

405 బిలియన్ పారామీటర్లతో, ల్లామా 3.1 ఓపెన్ సోర్స్ LLM విభాగంలో పడుతుంది. 

ఈ మోడల్ ముందటి విడుదల అయిన ల్లామా 3 కి కంటే అప్‌గ్రేడ్ రూపంలో: 

- పెద్ద కాంటెక్ట్స్ విండో - 128k టోకెన్లు 8k టోకెన్ల కంటే
- పెద్ద గరిష్ట అవుట్‌పుట్ టోకెన్లు - 4096 vs 2048
- మెరుగైన బహుభాషా మద్దతు - శిక్షణ టోకెన్ల పెరుగుదల వల్ల

ఇవి ల్లామా 3.1 తో GenAI అప్లికేషన్లు తయారుచేసేటప్పుడు క్లిష్టమైన వినియోగాల్ని నిర్వహించడానికి సహాయం చేస్తాయి: 
- నేటివ్ ఫంక్షన్ కాలింగ్ - LLM వర్క్‌ఫ్లోకి బయట ఉన్న బాహ్య టూల్స్ మరియు ఫంక్షన్‌లను పిలవగల సామర్థ్యం
- మెరుగైన RAG ప్రదర్శన - పెద్ద కాంటెక్ట్స్ విండో కారణంగా
- సింథటిక్ డేటా జనరేషన్ - ఫైన్-ట్యూనింగ్ వంటి పనుల కోసం సమర్థవంతమైన డేటాను సృష్టించే సామర్థ్యం



### ఫంక్షన్ స్థానిక కాలింగ్  

Llama 3.1 ఫంక్షన్ లేదా టూల్ కాల్స్ చేయడంలో మరింత సమర్థవంతంగా ఉండేందుకు ఫైన్-ట్యూనింగ్ చేయబడింది. ఇది మరో రెండు అంతర్ప్రతిష్టిత టూల్స్ కలిగి ఉంది, ఇవి మోడల్ ఉపయోగించిన ప్రశ్న ఆధారంగా ఉపయోగించవలసిందిగా గుర్తించగలదు. ఈ టూల్స్ ఈ విధంగా ఉన్నాయి:  

- **బ్రేవ్ సెర్చ్** - వెబ్ సెర్చ్ చేయడం ద్వారా తాజా సమాచారాన్ని పొందడానికి ఉపయోగించవచ్చు, ఉదాహరణకు వాతావరణం వంటి సమాచారం  
- **వోల్ఫ్రామ్ అల్ఫా** - మరింత క్లిష్టమైన గణిత లెక్కింపుల కోసం ఉపయోగించవచ్చు, కనుక మీ స్వంత ఫంక్షన్లు రాయడం అవసరం లేదు.  

మీరు మీ స్వంత కస్టమ్ టూల్స్ కూడా సృష్టించవచ్చు, వాటిని LLM కాల్ చేయగలదు.  

క్రింది కోడ్ ఉదాహరణలో:  

- అందుబాటులో ఉన్న టూల్స్ (brave_search, wolfram_alpha) ని సిస్టమ్ ప్రాంప్ట్‌లో నిర్వచించాము.  
- ఒక యూజర్ ప్రాంప్ట్ పంపించి, ఒక నిర్దిష్ట నగరంలో వాతావరణం గురించి అడుగుతారు.  
- LLM బ్రేవ్ సెర్చ్ టూల్ కాల్‌తో స్పందిస్తుంది, ఇది ఇలా కనిపిస్తుంది `<|python_tag|>brave_search.call(query="Stockholm weather")`  

*గమనిక: ఈ ఉదాహరణలో కేవలం టూల్ కాల్ చేయబడుతుంది, ఫలితాల‌ను పొందాలనుకుంటే మీరు బ్రేవ్ API పేజీపై ఉచిత ఖాతాను సృష్టించి ఆ ఫంక్షన్‌ను నిర్వచించవలసి ఉంటుంది`  


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### లామా 3.2

లాంగ్వేజ్ మోడల్ అయినప్పటికీ, లామా 3.1కు ఉన్న ఒక పరిమితి మల్టీమోడాలిటీ. అంటే, చిత్రాలు వంటి వివిధ రకాల ఇన్‌పుట్‌లను ప్రాంప్ట్‌లుగా ఉపయోగించడం మరియు ప్రతిస్పందనలు అందించడం. ఈ సామర్థ్యం లామా 3.2 యొక్క ప్రధాన ఫీచర్లలో ఒకటి. ఈ ఫీచర్లు కూడా ఉన్నాయి:

- మల్టీమోడాలిటీ - టెక్స్ట్ మరియు ఇమేజ్ ప్రాంప్ట్‌లు రెండింటినీ మూల్యాంకనం చేసే సామర్థ్యం ఉంది
- చిన్న నుండి మధ్యస్థ పరిమాణ వేరియేషన్లు (11B మరియు 90B) - ఇది సౌలభ్యవంతమైన డిప్లాయ్‌మెంట్ ఆప్షన్లను అందిస్తుంది,
- టెక్స్ట్-ఒన్‌లీ వేరియేషన్లు (1B మరియు 3B) - ఇది మోడల్‌ను ఎడ్జ్ / మొబైల్ డివైసులపై డిప్లాయ్ చేయడానికి అనుమతించి తక్కువ లేటెన్సీని అందిస్తుంది

మల్టీమోడల్ సపోర్ట్ ఓపెన్ సోర్స్ మోడల్స్ ప్రపంచంలో పెద్ద అడుగు. క్రింద ఉన్న కోడ్ ఉదాహరణలో లామా 3.2 90B నుండి చిత్ర విశ్లేషణ కోసం చిత్రాన్ని మరియు టెక్స్ట్ ప్రాంప్ట్‌ను రెండింటినీ తీసుకుంటుంది.

### లామా 3.2తో మల్టీమోడల్ సపోర్ట్


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
