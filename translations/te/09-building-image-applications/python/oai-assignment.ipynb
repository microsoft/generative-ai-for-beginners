## DALL-E మరియు Midjourney అంటే ఏమిటి? 

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) మరియు [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) అనేవి రెండు అత్యంత ప్రాచుర్యంలో ఉన్న ఇమేజ్ జనరేషన్ మోడల్స్, అవి ప్రాంప్ట్‌లను ఉపయోగించి ఇమేజ్‌లు ఉత్పత్తి చేయగలవు.

### DALL-E

DALL-E తో ప్రారంభిద్దాం, ఇది పాఠ్య వివరణల నుండి ఇమేజ్‌లు సృష్టించే జనరేటివ్ AI మోడల్.

> [DALL-E రెండు మోడల్స్, CLIP మరియు diffused attention కలయిక](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **CLIP**, ఇది ఇమేజ్‌లు మరియు పాఠ్యాల నుండి డేటా యొక్క సంఖ్యాత్మక ప్రాతినిధ్యాలు, అంటే ఎంబెడింగ్స్‌ని ఉత్పత్తి చేసే మోడల్.  

- **Diffused attention**, ఇది ఎంబెడింగ్స్ నుండి ఇమేజ్‌లను సృష్టించే మోడల్. DALL-E ఇమేజ్‌లు మరియు పాఠ్యాల డేటాసెట్‌పై శిక్షణ పొందినది మరియు పాఠ్య వివరణల నుండి ఇమేజ్‌లను ఉత్పత్తి చేయవచ్చు. ఉదాహరణకు, DALL-E ఒక టోపీ పెట్టుకున్న పిల్లి లేదా మొహాక్ గల కుక్క యొక్క ఇమేజ్‌లను సృష్టించడానికీ ఉపయోగపడుతుంది.

Midjourney కూడా DALL-Eలా పనిచేస్తుంది, ఇది పాఠ్య ప్రాంప్ట్‌ల నుండి ఇమేజ్‌లు సృష్టిస్తుంది. Midjourney కూడా “టోపీ పెట్టుకున్న పిల్లి” లేదా “మోహాక్ గల కుక్క” వంటి ప్రాంప్ట్‌లను ఉపయోగించి ఇమేజ్‌లను ఉత్పత్తి చేయవచ్చు.

## DALL-E మరియు Midjourney ఎలా పని చేస్తాయి 

మొదట [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E అనేది ట్రాన్స్‌ఫార్మర్ వికారంపై ఆధారపడిన జనరేటివ్ AI మోడల్ మరియు *ఆటోరెగ్రెసివ్ ట్రాన్స్‌ఫార్మర్* కలిగి ఉంటుంది.  

ఒక *ఆటోరెగ్రెసివ్ ట్రాన్స్‌ఫార్మర్* ఎలా పాఠ్య వివరణల నుండి ఇమేజ్‌లను సృష్టిస్తుందని నిర్వచిస్తుంది, ఇది ఒక సమయంలో ఒక్క పిక్సెల్‌ని సృష్టించి, ఆ సృష్టించబడిన పిక్సెల్‌లను ఉపయోగించి తదుపరి పిక్సెల్‌ని సృష్టిస్తుంది. న్యూరల్ నెట్‌వర్క్‌లో పలువురు లేయర్‌ల ద్వారా గుండా వెళ్లి, చివరికి పూర్తి చిత్రం వస్తుంది.  

ఈ ప్రక్రియతో, DALL-E, సృష్టించే చిత్రంలో లక్షణాలు, వస్తువులు, గుణాలు మరియు మరెన్నో అంశాలను నియంత్రిస్తుంది. అయినప్పటికీ, DALL-E 2 మరియు 3 సృష్టించిన చిత్రంపై మరింత నియంత్రణ కలిగి ఉంటాయి, 


## మీ మొదటి చిత్రం సృష్టి అప్లికేషన్ ను నిర్మించడం

కాబట్టి చిత్రం సృష్టి అప్లికేషన్ ని నిర్మించడానికి ఏమి అవసరం? మీరు ఈ క్రింది లైబ్రరీలను ఉపయోగించాలి:

- **python-dotenv**, ఈ లైబ్రరీని మీ రహస్యాలు కోడ్ నుండి దూరంగా *.env* ఫైలులో ఉంచడానికి చాలా సిఫార్సు చేయబడింది.
- **openai**, ఈ లైబ్రరీ OpenAI APIతో ఇంటరాక్ట్ చేయడానికి మీరు ఉపయోగిస్తారు.
- **pillow**, పఠాన్ లో చిత్రాలతో పని చేయడానికి.
- **requests**, HTTP అభ్యర్థనలు చేయడానికి సహాయం చేసుతుంది.


1. క్రింది కంటెంట్ తో *.env* అనే ఫైలును సృష్టించండి:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. పై లైబ్రరీలను *requirements.txt* అనే ఫైల్‌లో ఇలా సేకరించండి:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. తర్వాత, వర్చువల్ ఎన్విరాన్మెంట్‌ను సృష్టించి లైబ్రరీలను ఇన్స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> విండోస్ కోసం, మీ వర్చువల్ ఎన్విరన్మెంట్‌ని సృష్టించడానికి మరియు యాక్టివేట్ చేయడానికి క్రింది కమాండ్లను ఉపయోగించండి:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* అనే ఫైల్‌లో క్రింది కోడ్‌ను జోడించండి:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ను ఇంపోర్ట్ చేయండి
    dotenv.load_dotenv()

    # OpenAI ఆబ్జెక్ట్ సృష్టించండి
    client = openai.OpenAI()


    try:
        # చిత్రం జనరేషన్ API ఉపయోగించి చిత్రం సృష్టించండి
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
            size='1024x1024',
            n=1
        )
        # స్టోర్ చేసిన చిత్రానికి డైరెక్టరీ సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')

        # డైరెక్టరీ లేకపోతే, దాన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # చిత్రం పథాన్ని ఆరంభించండి (ఫైల్టైపు png ఉండాలని గమనించండి)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # జనరేట్ చేసిన చిత్రాన్ని పొందండి
        print(generation_response)

        image_url = generation_response.data[0].url # స్పందన నుండి చిత్రం URL ను తీసుకోండి
        generated_image = requests.get(image_url).content  # చిత్రాన్ని డౌన్లోడ్ చేయండి
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # డిఫాల్ట్ చిత్రం వీక్షకలో చిత్రాన్ని ప్రదర్శించండి
        image = Image.open(image_path)
        image.show()

    # తప్పిదాలను క్యాచ్చి తీసుకోండి
    except openai.BadRequestError as err:
        print(err)

    ```

- మొట్టమొదట, మేము అవసరమైన లైబ్రరీలని ఇంపోర్ట్ చేసుకుంటాము, ఇందులో OpenAI లైబ్రరీ, dotenv లైబ్రరీ, requests లైబ్రరీ, మరియు Pillow లైబ్రరీ ఉన్నాయి.

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv 
    ```

- ఆ తర్వాత, మేము ఆబ్జెక్ట్ సృష్టిస్తాము, ఇది మీ ``.env`` నుండి API కీని పొందుతుంది.

    ```python

    ```

- తర్వాత, మేము చిత్రం సృష్టిస్తాము:

    ```python

        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',

    )
    ```

    పై కోడ్ ఒక JSON ఆబ్జెక్ట్తో స్పందిస్తుంది, దీనిలో సృష్టించిన చిత్రం యొక్క URL ఉంటుంది. మేము ఆ URL ఉపయోగించి చిత్రాన్ని డౌన్లోడ్ చేసి ఫైల్‌గా సేవ్ చేసుకోవచ్చు.

- చివరిగా, మేము చిత్రాన్ని తెరిచి సాధారణ ఇమేజ్ వ్యూ‌వర్‌తో ప్రదర్శిస్తాము:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### చిత్రం సృష్టించడంపై మరిన్ని వివరాలు

చిత్రాన్ని సృష్టించే కోడ్‌ను మరింత స్పష్టంగా చూద్దాం:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',
        n=1
    )
```

- **prompt**, చిత్రం సృష్టించడానికి ఉపయోగించే టెక్స్ట్ ప్రాంప్ట్. ఈ సందర్భంలో, మేము "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" అనే ప్రాంప్ట్‌ని ఉపయోగిస్తున్నాము.
- **size**, సృష్టించే చిత్రం యొక్క పరిమాణం. ఈ సందర్భంలో, మేము 1024x1024 పిక్సెల్స్ పరిమాణంలో చిత్రం సృష్టిస్తున్నాము.
- **n**, సృష్టించే చిత్రాల సంఖ్య. ఈ సందర్భంలో, మేము రెండు చిత్రాలను సృష్టిస్తున్నాము.

మరిన్ని చిత్రాలతో చేయగల విషయాలని మేము తదుపరి విభాగంలో చర్చిస్తాము.

## చిత్ర సృష్టించడంలో అదనపు నైపుణ్యాలు

ఇప్పటివరకు మీరు చూచినట్లుగా, కొన్ని Python లైన్ల ద్వారా మేము చిత్రం సృష్టించగలిగాము. అయినప్పటికీ, చిత్రాలతో ఇంకా ఎన్నో చేయవచ్చు.

మీరు క్రింది పనులు కూడా చేయవచ్చు:

- **సవరించుకోండి**. ఒక ఉన్న చిత్రం, దాని మాస్క్ మరియు ప్రాంప్ట్ ఇచ్చి, మీరు చిత్రాన్ని మార్చవచ్చు. ఉదాహరణకు, మీరు చిత్రం యొక్క ఒక భాగానికి ఏదైనా జోడించవచ్చు. మన బన్నీ చిత్రంగా ఊహించండి, మీరు బన్నీకి టోపీ జోడించవచ్చు. అది ఎలా చేయాలో అంటే, చిత్రం, మాస్క్ (మార్పు చేయవలసిన ప్రాంతాన్ని గుర్తించే) మరియు టెక్స్ట్ ప్రాంప్ట్ ఇవ్వడం.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ప్రాథమిక చిత్రం పక్కన మాత్రమేrabbit ఉంటుంది కానీ తుది చిత్రంలో బన్నీకి టోపీ ఉంటుంది.
    
- **వేరియేషన్లు సృష్టించండి**. ఆలోచన ఏమిటంటే, మీరు ఒక ఉన్న చిత్రాన్ని తీసుకుని దానికి వేరియేషను చేయాలని అడగండి. వేరియేషన్ సృష్టించడానికి, మీరు ఒక చిత్రం మరియు టెక్స్ట్ ప్రాంప్ట్ ఇచ్చి క్రింది విధంగా కోడ్ రాయండి:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
