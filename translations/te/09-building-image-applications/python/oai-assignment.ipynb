## మీ మొదటి చిత్రం సృష్టి అప్లికేషన్ నిర్మాణం

అప్పుడు చిత్రం సృష్టి అప్లికేషన్ నిర్మించడానికి ఏమి అవసరం? మీరు క్రింది లైబ్రరీలను అవసరం:

- **python-dotenv**, మీ రహస్యాలను కోడ్ నుండి దూరంగా *.env* ఫైల్‌లో ఉంచడానికి ఈ లైబ్రరీని ఉపయోగించడం మిమ్మల్ని గట్టిగా సిఫార్సు చేస్తారు.
- **openai**, ఈ లైబ్రరీని మీరు OpenAI APIతో ఇంటరాక్ట్ చేయడానికి ఉపయోగిస్తారు.
- **pillow**, Pythonలో చిత్రాలతో పని చేయడానికి.
- **requests**, HTTP అభ్యర్థనలు చేయడంలో మీకు సహాయం చేయడానికి.


1. క్రింది విషయంతో *.env* ఫైల్ సృష్టించండి:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```


1. పై లైబ్రరీలను *requirements.txt* అనే ఫైల్‌లో ఈ విధంగా సేకరించండి:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. తరువాత, వర్చువల్ ఎన్విరాన్‌మెంట్ సృష్టించి లైబ్రరీలను ఇన్‌స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows కోసం, మీ వర్చువల్ ఎన్విరాన్‌మెంట్‌ను సృష్టించడానికి మరియు యాక్టివేట్ చేయడానికి క్రింది కమాండ్లను ఉపయోగించండి:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* అనే ఫైల్‌లో క్రింది కోడ్‌ను జోడించండి:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ను దిగుమతి చేసుకోండి
    dotenv.load_dotenv()

    # OpenAI ఆబ్జెక్ట్‌ను సృష్టించండి
    client = OpenAI()


    try:
        # ఇమేజ్ జనరేషన్ API ఉపయోగించి చిత్రం సృష్టించండి
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
            size='1024x1024',
            n=1
        )
        # నిల్వచేసిన చిత్రానికి డైరెక్టరీని సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')

        # డైరెక్టరీ లేకపోతే, దాన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # చిత్రం మార్గాన్ని ప్రారంభించండి (ఫైల్ టైపు png కావాలి)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # సృష్టించిన చిత్రాన్ని పొందండి
        print(generation_response)

        image_url = generation_response.data[0].url # ప్రతిస్పందన నుండి చిత్రం URL ను తీసుకోండి
        generated_image = requests.get(image_url).content  # చిత్రాన్ని డౌన్లోడ్ చేయండి
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # డిఫాల్ట్ చిత్రం వీక్షకంలో చిత్రాన్ని ప్రదర్శించండి
        image = Image.open(image_path)
        image.show()

    # తప్పిదాలను పట్టుకోండి
    except client.error.InvalidRequestError as err:
        print(err)

    ```

    ```python
    )
    ```

    పై కోడ్ ఒక JSON ఆబ్జెక్టుతో స్పందిస్తుంది, ఇది సృష్టించిన చిత్ర URLని కలిగి ఉంటుంది. మేము ఆ URLని ఉపయోగించి చిత్రాన్ని డౌన్లోడ్ చేసి ఫైల్‌గా సేవ్ చేయవచ్చు.

- చివరగా, మేము చిత్రాన్ని తెరిచి సాధారణ చిత్రం వీక్షకుడిని ఉపయోగించి దాన్ని ప్రదర్శిస్తాము:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### చిత్రాన్ని సృష్టించడంపై మరిన్ని వివరాలు

చిత్రాన్ని సృష్టించే కోడ్‌ను మరింత వివరంగా చూద్దాం:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',
        n=1
    )
```

- **prompt**, చిత్రం సృష్టించడానికి ఉపయోగించే టెక్స్ట్ ప్రాంప్ట్. ఈ సందర్భంలో, మేము "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" అనే ప్రాంప్ట్‌ను ఉపయోగిస్తున్నాము.
- **size**, సృష్టించబడే చిత్ర పరిమాణం. ఈ సందర్భంలో, మేము 1024x1024 పిక్సెల్స్ పరిమాణంలో చిత్రం సృష్టిస్తున్నాము.
- **n**, సృష్టించబడే చిత్రాల సంఖ్య. ఈ సందర్భంలో, మేము రెండు చిత్రాలను సృష్టిస్తున్నాము.

చిత్రాలతో మీరు చేయగల మరిన్ని విషయాలు ఉన్నాయి, వాటిని మేము తదుపరి విభాగంలో చర్చిస్తాము.

## చిత్రం సృష్టించడంలో అదనపు సామర్థ్యాలు

ఇప్పటివరకు మీరు కొద్ది పంక్తుల Python కోడ్ ఉపయోగించి చిత్రం ఎలా సృష్టించగలమో చూశారు. అయితే, చిత్రాలతో మీరు చేయగల మరిన్ని విషయాలు ఉన్నాయి.

మీరు క్రింది పనులు కూడా చేయవచ్చు:

- **సవరింపులు చేయండి**. ఇప్పటికే ఉన్న చిత్రానికి మాస్క్ మరియు ప్రాంప్ట్‌ను అందించడం ద్వారా, మీరు చిత్రాన్ని మార్చవచ్చు. ఉదాహరణకు, మీరు చిత్రంలోని ఒక భాగానికి ఏదైనా జోడించవచ్చు. మన బన్నీ చిత్రాన్ని ఊహించండి, మీరు బన్నీకి టోపీ జోడించవచ్చు. మీరు ఎలా చేస్తారు అంటే, చిత్రం, మాస్క్ (మార్పు చేయవలసిన ప్రాంతాన్ని గుర్తించే) మరియు టెక్స్ట్ ప్రాంప్ట్‌ను అందించడం ద్వారా.

    ```python
    response = openai.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    ప్రాథమిక చిత్రం కేవలం బన్నీని మాత్రమే కలిగి ఉంటుంది కానీ తుది చిత్రంలో బన్నీపై టోపీ ఉంటుంది.
    
- **వేరియేషన్లు సృష్టించండి**. ఆలోచన ఏమిటంటే, మీరు ఇప్పటికే ఉన్న చిత్రాన్ని తీసుకుని వేరియేషన్లు సృష్టించాలని అడుగుతారు. వేరియేషన్ సృష్టించడానికి, మీరు చిత్రం మరియు టెక్స్ట్ ప్రాంప్ట్‌ను అందించి క్రింది విధంగా కోడ్ రాస్తారు:

    ```python
    response = openai.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
