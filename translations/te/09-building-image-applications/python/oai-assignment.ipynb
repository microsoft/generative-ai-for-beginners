# చిత్ర నిర్మాణ అనువర్తనాలు (OpenAI)

LLMs కేవలం టెక్స్ట్ తయారుచేయడమే కాదు, మీరు టెక్స్ట్ వివరణల నుండి చిత్రాలు కూడా erzeugen చేసుకోవచ్చు. చిత్రాలు ఒక మోడ్ గా MedTech, ఆర్కిటెక్చర్, టూరిజం, గేమ్ అభివృద్ధి, మార్కెటింగ్ మరియు మరిన్ని రంగాలలో ఉపయోగకరంగా ఉంటాయి. ఈ పాఠంలో మనం OpenAI వేదికపై నేటి **GPT Image** మోడల్స్ తో పని చేస్తాము.

## నేర్చుకునే లక్ష్యాలు

ఈ పాఠం ముగింపు నాటికి మీరు చేయగలుగుతారు:

- చిత్ర నిర్మాణం అంటే ఏమిటి మరియు ఎక్కడ ఉపయోగపడుతుంది అనేది వివరించగలగడం.
- `gpt-image` మోడల్ కుటుంబాన్ని అర్థం చేసుకోవడం మరియు అది పూర్వపు DALL·E మోడల్స్ నుండి ఎలా భిన్నమో తెలుసుకోవడం.
- ఒక చిత్ర నిర్మాణ అనువర్తనం తయారుచేసి చిత్రాలను ఎడిట్ చేయడం.

## చిత్ర నిర్మాణం అంటే ఏమిటి?

చిత్ర నిర్మాణ మోడల్స్ ఒక టెక్స్ట్ ప్రాంప్ట్ నుండి చిత్రాలను సృష్టిస్తాయి. ఆధునిక మోడల్స్ వంటి `gpt-image` శిక్షణ సమయంలో టెక్స్ట్ మరియు చిత్రాల మధ్య సంబంధాన్ని నేర్పుకుని, ఆ తరువాత పునరావృతంగా యాదృచ్ఛిక శబ్దాన్ని తీసుకొని మీ ప్రాంప్ట్ కు సరిపోయే చిత్రంగా మార్చుతాయి.

రెండు ప్రసిద్ధ చిత్ర మోడల్ కుటుంబాలు:

- **`gpt-image` (OpenAI)** — ఇక్కడ ఈ పాఠంలో ఉపయోగించిన ప్రస్తుత తరం. ఇది టెక్స్ట్-टు-చిత్రం మరియు చిత్ర ఎడిటింగ్ (మాస్క్‌తో inpainting) ని మద్దతు ఇస్తుంది.
- **Midjourney** — ఒక ప్రముఖ మూడో కెంపెనీ మోడల్, దాని స్వంత సేవ మరియు Discord ఆధారిత వర్క్‌ఫ్లో తో.

> పాత OpenAI చిత్ర మోడల్స్ — **DALL·E 2** మరియు **DALL·E 3** — ప్రాచీన మోడల్స్. DALL·E 3 కొత్త అమల్లోకి ధరింపులకు ఇక అందుబాటులో లేదు, మరియు `create_variation` వంటి ఫీచర్లు కేవలం DALL·E 2 లో ఉన్నాయి. కొత్త అనువర్తనాల కొరకు `gpt-image` మోడల్స్ ను ఉపయోగించండి.

> **గమనిక:** `gpt-image` మోడల్స్ ఉత్పత్తి చేసిన చిత్రాన్ని **base64** (`b64_json`) రూపంలో ఇస్తాయి, URL రూపంలో కాదు. మీ కోడ్ base64 స్ట్రింగ్ ను బైట్స్ గా డీకోడ్ చేసి సేవ్ చేస్తుంది — డౌన్లోడ్ చేసుకునే చిత్రం URL లేదు.


## మీ మొదటి చిత్రం ఉత్పత్తి అప్లికేషన్‌ని నిర్మించడం

కాబట్టి చిత్రం ఉత్పత్తి అప్లికేషన్‌ని నిర్మించడానికి ఏమి అవసరం? మీరు క్రింది లైబ్రరీలను అవసరం:

- **python-dotenv**, మీ రహస్యాలను కోడ్ నుండి దూరంగా *.env* ఫైల్‌లో ఉంచడానికి ఈ లైబ్రరీని ఉపయోగించడం ఎంతో సిఫార్సు చేయబడుతుంది.
- **openai**, ఈ లైబ్రరీని OpenAI APIతో ఇంటరాక్ట్ అయ్యేందుకు మీరు ఉపయోగిస్తారు.
- **pillow**, Pythonలో చిత్రాలతో పని చేయడానికి.


1. క్రింది సంగతులతో ఒక *.env* ఫైల్ సృష్టించండి:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. పై లైబ్రరీలను *requirements.txt* అనే ఫైల్‌లో ఈ విధంగా సేకరించండి:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. తరువాత, వర్చువల్ סביבన్మెంట్ సృష్టించి లైబ్రరీలను ఇన్‌స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> విండోస్ కోసం, మీ వర్చువల్ ఎన్విరాన్‌మెంట్‌ను సృష్టించడానికి మరియు సక్రియం చేయడానికి క్రింది కమాండ్లను ఉపయోగించండి:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* అనే ఫైల్‌లో క్రింది కోడ్‌ను జోడించండి:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv ను దిగుమతించండి
    dotenv.load_dotenv()

    # OpenAI ఆబ్జెక్ట్‌ను సృష్టించండి (.env నుండి OPENAI_API_KEY ని చదువుతుంది)
    client = openai.OpenAI()


    try:
        # ఇమేజ్ జనరేషన్ API ఉపయోగించి చిత్రాన్ని సృష్టించండి
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ చేర్చండి
            size='1024x1024',
            n=1
        )
        # నిల్వ చేయబడిన చిత్రానికి డైరెక్టరీని సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')

        # డైరెక్టరీ ఉండనప్పుడు, దాన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # చిత్ర మార్గాన్ని ప్రారంభించండి (ఫైల్ టైప్ png అవ్వాలి గమనించండి)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image మోడల్స్ చిత్రం base64 (b64_json) రూపంలో ఇస్తాయి, URL రూపంలో కాదు
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # డిఫాల్ట్ చిత్రం వీవర్ లో చిత్రాన్ని ప్రదర్శించండి
        image = Image.open(image_path)
        image.show()

    # తప్పిదాలను పట్టుకోండి
    except openai.BadRequestError as err:
        print(err)

    ```

ఈ కోడ్‌ను వివరించుకుందాం:

- మొదటిగా, మనం అవసరమైన లైబ్రరీలను దిగుమతి చేసుకుంటాము, అందులో OpenAI లైబ్రరీ, dotenv లైబ్రరీ, base64 మాడ్యూల్ మరియు Pillow లైబ్రరీ ఉన్నాయి.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- అంతకు మించి, మనం క్లయింట్‌ను సృష్టిస్తాము, ఇది మీ ``.env`` నుండి API కీని చదువుకుంటుంది.

    ```python
    # OpenAI ఆబ్జెక్ట్‌ను సృష్టించండి
    client = openai.OpenAI()
    ```

- తరువాత, మనం చిత్రం సృష్టిస్తాము:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` మోడల్స్ చిత్రం ని **base64** స్ట్రింగ్‌గా `data[0].b64_json`లో ఇస్తాయి. మనం దాన్ని బైట్స్‌గా డికోడ్ చేసి ఫైల్‌గా రాస్తాము — డౌన్లోడ్ చేసుకునే URL లేదు.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- చివరగా, మనం చిత్రాన్ని తెరిచి పాతదైన చిత్ర వీక్షణ కర్తను ఉపయోగించి చూపిస్తాము:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### చిత్రాన్ని సృష్టించడంపై మరిన్ని వివరాలు

`images.generate` పారామితులని చూద్దాం:

- **model**, ఉపయోగించవలసిన చిత్రం మోడల్ (ఉదాహరణకు `gpt-image-1`).
- **prompt**, చిత్రాన్ని సృష్టించడానికి ఉపయోగించే వచన ప్రాంప్ట్. ఇక్కడ "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, సృష్టించబడిన చిత్రం పరిమాణం (ఉదాహరణకు `1024x1024`, `1536x1024`, `1024x1536`, లేదా `"auto"`).
- **n**, సృష్టించబడిన చిత్రాల సంఖ్య. ఇక్కడ ఒకటి సృష్టించాం.

> చిత్రం మోడల్స్ `temperature` పారామితి తీసుకోవు — అది వచన-సృష్టి నియంత్రణ. వైవిధ్యాన్ని పొందడానికి, API ను మళ్ళీ పిలవండి; వైవిధ్యాన్ని తగ్గించడానికి, మీ ప్రాంప్ట్‌ను మరింత నిర్దిష్టంగా చేయండి.

## చిత్రం సృష్టించడంలో అదనపు సామర్థ్యాలు

మీరు పాథాన్‌లో కొన్ని లైన్లతో ఎలా చిత్రం సృష్టించాలో చూసారు. `gpt-image` మోడల్స్ ఇప్పటికే ఉన్న చిత్రాన్ని **సవరించగలవు** కూడా. ఒక చిత్రం, ఐచ్ఛికం **మాస్క్** (మార్పు చేయాల్సిన ప్రాంతాన్ని గుర్తు చేసే) మరియు ప్రాంప్ట్ ఇచ్చి, మీరు చిత్రం ఒక భాగాన్ని మార్చవచ్చు — ఉదాహరణకు, మన బన్నీకి టోపీ వేసే విధంగా.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ఎడిట్లు కూడా base64గా తిరిగి పంపబడతాయి
edited_image = base64.b64decode(response.data[0].b64_json)
```

ప్రాథమిక చిత్రం కేవలం సిరాయి ని కలిగి ఉంది; చివరి చిత్రంలో సిరాయి పై టోపీ ఉంది.
