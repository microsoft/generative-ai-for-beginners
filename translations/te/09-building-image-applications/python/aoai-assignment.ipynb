## DALL-E మరియు Midjourney ఎలా పనిచేస్తాయి

ముందుగా, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E అనేది ట్రాన్స్‌ఫార్మర్ ఆర్కిటెక్చర్ ఆధారిత జనరేటివ్ AI మోడల్, ఇది *ఆటోరెగ్రెసివ్ ట్రాన్స్‌ఫార్మర్* కలిగి ఉంటుంది.

*ఆటోరెగ్రెసివ్ ట్రాన్స్‌ఫార్మర్* ఒక మోడల్ టెక్స్ట్ వివరణల నుండి చిత్రాలను ఎలా సృష్టిస్తుందో నిర్వచిస్తుంది, ఇది ఒక్కో పిక్సెల్‌ను ఒకటిగా సృష్టించి, ఆ పిక్సెల్‌లను ఉపయోగించి తదుపరి పిక్సెల్‌ను సృష్టిస్తుంది. ఇది న్యూరల్ నెట్‌వర్క్‌లో అనేక లేయర్ల ద్వారా పాస్ అవుతుంది, చివరికి చిత్రం పూర్తవుతుంది.

ఈ ప్రక్రియతో, DALL-E సృష్టించే చిత్రంలో లక్షణాలు, వస్తువులు, గుణాలు మరియు మరిన్ని నియంత్రిస్తుంది. అయితే, DALL-E 2 మరియు 3 సృష్టించిన చిత్రంపై మరింత నియంత్రణ కలిగి ఉంటాయి.


## మీ మొదటి చిత్రం సృష్టి అప్లికేషన్ నిర్మాణం

అప్పుడు చిత్రం సృష్టి అప్లికేషన్ నిర్మించడానికి ఏమి అవసరం? మీరు క్రింది లైబ్రరీలను అవసరం:

- **python-dotenv**, మీ రహస్యాలను కోడ్ నుండి దూరంగా *.env* ఫైల్‌లో ఉంచడానికి ఈ లైబ్రరీని ఉపయోగించమని మిమ్మల్ని గట్టిగా సిఫార్సు చేస్తారు.
- **openai**, ఈ లైబ్రరీని మీరు OpenAI APIతో ఇంటరాక్ట్ చేయడానికి ఉపయోగిస్తారు.
- **pillow**, Pythonలో చిత్రాలతో పని చేయడానికి.
- **requests**, HTTP అభ్యర్థనలు చేయడంలో సహాయపడుతుంది.

1. క్రింది విషయాలతో *.env* ఫైల్ సృష్టించండి:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    ```

    మీ వనరుకు Azure పోర్టల్‌లో "Keys and Endpoint" విభాగంలో ఈ సమాచారాన్ని కనుగొనండి.


1. పై లైబ్రరీలను *requirements.txt* అనే ఫైల్‌లో ఈ విధంగా సేకరించండి:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. తరువాత, వర్చువల్ ఎన్విరాన్‌మెంట్ సృష్టించి లైబ్రరీలను ఇన్‌స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows కోసం, మీ వర్చువల్ ఎన్విరాన్‌మెంట్‌ను సృష్టించడానికి మరియు యాక్టివేట్ చేయడానికి క్రింది కమాండ్లను ఉపయోగించండి:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* అనే ఫైల్‌లో క్రింది కోడ్‌ను జోడించండి:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # dotenv ను దిగుమతి చేసుకోండి
    dotenv.load_dotenv()
    
    # ఎండ్పాయింట్ మరియు కీని పర్యావరణ చరాలు నుండి పొందండి
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # API సంస్కరణను కేటాయించండి (DALL-E ప్రస్తుతం 2023-06-01-preview API సంస్కరణకు మాత్రమే మద్దతు ఇస్తుంది)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # చిత్రం సృష్టించడానికి చిత్రం ఉత్పత్తి API ను ఉపయోగించండి
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # నిల్వచేసిన చిత్రానికి డైరెక్టరీని సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')
    
        # డైరెక్టరీ లేకపోతే, దాన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # చిత్రం మార్గాన్ని ప్రారంభించండి (ఫైల్ రకం png కావాలి)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # ఉత్పత్తి చేసిన చిత్రాన్ని పొందండి
        image_url = generation_response["data"][0]["url"]  # ప్రతిస్పందన నుండి చిత్రం URL ను తీసుకోండి
        generated_image = requests.get(image_url).content  # చిత్రాన్ని డౌన్లోడ్ చేయండి
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # డిఫాల్ట్ చిత్రం వీక్షకంలో చిత్రాన్ని ప్రదర్శించండి
        image = Image.open(image_path)
        image.show()
    
    # తప్పిదాలను పట్టుకోండి
    except openai.InvalidRequestError as err:
        print(err)

    ```

ఈ కోడ్‌ను వివరించుకుందాం:

    ```python
    import dotenv
    ```

- తరువాత, మేము *.env* ఫైల్ నుండి ఎన్విరాన్‌మెంట్ వేరియబుల్స్‌ను లోడ్ చేస్తాము.

    ```python
    # డాట్‌ఎన్‌వి దిగుమతి చేసుకోండి
    dotenv.load_dotenv()
    ```

- ఆ తర్వాత, మేము OpenAI API కోసం ఎండ్‌పాయింట్, కీ, వెర్షన్ మరియు టైప్‌ను సెట్ చేస్తాము.

    ```python
    # ఎండ్పాయింట్ మరియు కీని వాతావరణ చరాలు నుండి పొందండి
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY'] 

    # వెర్షన్ మరియు రకాన్ని జోడించండి, Azure ప్రత్యేకమైనది
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    ```

- తరువాత, మేము చిత్రం సృష్టిస్తాము:

    ```python
    # ఇమేజ్ జనరేషన్ API ఉపయోగించి ఒక చిత్రం సృష్టించండి
    generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',
        n=2,
        temperature=0,
    )
    ```

    పై కోడ్ ఒక JSON ఆబ్జెక్ట్‌తో స్పందిస్తుంది, ఇది సృష్టించిన చిత్ర URL‌ను కలిగి ఉంటుంది. మేము ఆ URL ఉపయోగించి చిత్రాన్ని డౌన్లోడ్ చేసి ఫైల్‌గా సేవ్ చేయవచ్చు.

- చివరగా, మేము చిత్రాన్ని తెరిచి, సాధారణ చిత్రం వీక్షకుడిని ఉపయోగించి దాన్ని ప్రదర్శిస్తాము:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### చిత్రాన్ని సృష్టించడంపై మరిన్ని వివరాలు

చిత్రాన్ని సృష్టించే కోడ్‌ను మరింత వివరంగా చూద్దాం:

```python
generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ఇక్కడ మీ ప్రాంప్ట్ టెక్స్ట్‌ను నమోదు చేయండి
        size='1024x1024',
        n=2,
        temperature=0,
    )
```

- **prompt**, చిత్రం సృష్టించడానికి ఉపయోగించే టెక్స్ట్ ప్రాంప్ట్. ఈ సందర్భంలో, మేము "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" అనే ప్రాంప్ట్‌ను ఉపయోగిస్తున్నాము.
- **size**, సృష్టించబడే చిత్ర పరిమాణం. ఈ సందర్భంలో, మేము 1024x1024 పిక్సెల్స్ పరిమాణంలో చిత్రాన్ని సృష్టిస్తున్నాము.
- **n**, సృష్టించబడే చిత్రాల సంఖ్య. ఈ సందర్భంలో, మేము రెండు చిత్రాలను సృష్టిస్తున్నాము.
- **temperature**, ఇది జనరేటివ్ AI మోడల్ అవుట్‌పుట్ యొక్క యాదృచ్ఛికతను నియంత్రించే పారామీటర్. టెంపరేచర్ 0 నుండి 1 మధ్య విలువ, ఇక్కడ 0 అంటే అవుట్‌పుట్ నిర్దిష్టంగా ఉంటుంది మరియు 1 అంటే అవుట్‌పుట్ యాదృచ్ఛికంగా ఉంటుంది. డిఫాల్ట్ విలువ 0.7.

చిత్రాలతో మీరు చేయగల మరిన్ని విషయాలు ఉన్నాయి, వాటిని మేము తదుపరి విభాగంలో కవర్ చేస్తాము.

## చిత్రం సృష్టించడంలో అదనపు సామర్థ్యాలు

ఇప్పటివరకు మీరు చూసినట్లుగా, Python లో కొన్ని లైన్లతోనే మేము చిత్రం సృష్టించగలిగాము. అయితే, చిత్రాలతో మీరు చేయగల మరిన్ని విషయాలు ఉన్నాయి.

మీరు క్రింది పనులు కూడా చేయవచ్చు:

- **సవరింపులు చేయండి**. ఇప్పటికే ఉన్న చిత్రానికి మాస్క్ మరియు ప్రాంప్ట్‌ను అందించడం ద్వారా, మీరు చిత్రాన్ని మార్చవచ్చు. ఉదాహరణకు, చిత్రంలోని ఒక భాగానికి ఏదైనా జోడించవచ్చు. మన బన్నీ చిత్రాన్ని ఊహించండి, మీరు బన్నీకి టోపీ జోడించవచ్చు. మీరు ఎలా చేస్తారు అంటే, చిత్రం, మాస్క్ (మార్పు చేయవలసిన ప్రాంతాన్ని గుర్తించే) మరియు టెక్స్ట్ ప్రాంప్ట్‌ను అందించడం ద్వారా.

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    ప్రాథమిక చిత్రం కేవలం బన్నీని మాత్రమే కలిగి ఉంటుంది కానీ తుది చిత్రం బన్నీపై టోపీతో ఉంటుంది.
