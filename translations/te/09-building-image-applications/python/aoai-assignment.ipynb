# చిత్రం సృష్టి అనువర్తనాలను నిర్మించడం (Azure OpenAI)

LLMలు కేవలం వాక్య ఉత్పత్తి చేయడమే కాదు. మీరు టెక్స్ట్ వివరణల నుండి చిత్రాలు కూడా ఉత్పత్తి చేయవచ్చు. చిత్రాలు ఒక మోដాలిటీగా మెడ్‌టెక్, ఆర్కిటెక్చర్, టూరిజం, గేమ్ అభివృద్ధి, మార్కెటింగ్ మరియు మరిన్ని రంగాలలో ఉపయోగకరంగా ఉంటాయి. ఈ పాఠంలో మేము Microsoft Foundryలోని ఈరోజు **GPT చిత్రం** మోడళ్లతో పని చేస్తాము.

## నేర్చుకునే లక్ష్యాలు

ఈ పాఠం చివరలో మీరు చేయగలుగుతారు:

- చిత్రం సృష్టి అంటే ఏమిటి మరియు ఇది ఎక్కడ ఉపయోగకరం అనేది వివరించండి.
- `gpt-image` మోడల్ కుటుంబాన్ని అర్థం చేసుకోండి మరియు ఇది పూర్వపు DALL·E మోడళ్లతో ఎలా భిన్నంగా ఉందో తెలుసుకోండి.
- చిత్ర సృష్టి అనువర్తనం నిర్మించండి మరియు చిత్రాలను సంపాదించండి.

## చిత్రం సృష్టి అంటే ఏమిటి?

చిత్రం సృష్టి మోడళ్లు టెక్స్ట్ సూచన (prompt) నుండి చిత్రాలను సృష్టిస్తాయి. ఆధునిక మోడళ్లు `gpt-image` వంటి, శిక్షణ సమయంలో టెక్స్ట్ మరియు చిత్రాల మధ్య సంబంధాన్ని నేర్చుకుని, ఆపై యాదృచ్చిక శబ్దాన్ని సూచనతో సరిపోయే చిత్రంగా మలచుతాయి.

రెండు ప్రసిద్ధ చిత్రం మోడల్ కుటుంబాలు ఉన్నాయి:

- **`gpt-image` (OpenAI)** — ఈ పాఠంలో ఉపయోగిస్తున్న ప్రస్తుత తరం మోడల్. ఇది టెక్స్ట్-టు-చిత్ర సృష్టి మరియు చిత్రం ఎడిటింగ్ (మాస్క్‌తో ఇన్‌పైన్టింగ్) ను మద్దతు ఇస్తుంది.
- **Midjourney** — గలవైదు సంస్థతో ఉన్న ప్రాచుర్యం పొందిన మూడవ పార్శ్వ మోడల్ మరియు డిస్కార్డ్‌ ఆధారిత వర్క్‌ఫ్లో.

> పాత OpenAI చిత్రం మోడళ్లు — **DALL·E 2** మరియు **DALL·E 3** — పూర్వపు వర్గం. DALL·E 3 కొత్త నిర్వచనాల కోసం అందుబాటులో లేదు, మరియు `create_variation` వంటి ఫీచర్లు కేవలం DALL·E 2లో ఉన్నాయి. కొత్త అనువర్తనాలకు `gpt-image` మోడళ్లను ఉపయోగించండి.

Microsoft Foundryపై, **`gpt-image-2`** తాజా మరియు అత్యంత సమర్థవంతమైన చిత్రం మోడల్ మరియు సూచించబడిన మౌలికమైనది. `gpt-image-1.5` మరియు `gpt-image-1-mini` కూడా సాధారణంగా అందుబాటులో ఉన్నాయి.

> **గురుత్వము:** `gpt-image` మోడళ్లు ఉత్పత్తి చేసిన చిత్రాన్ని **base64** (`b64_json`) గా అందజేస్తాయి, URLగా కాదు. మీ కోడ్ base64 స్ట్రింగ్‌ని బైట్‌లుగా డీకోడ్ చేసి దాన్ని సంరక్షిస్తుంది — డౌన్‌లోడ్ చేసుకునేందుకు చిత్రం URL లేదు.


## మీ మొదటి ఇమేజ్ జనరేషన్ అప్లికేషన్‌ను నిర్మించడం

కాబట్టి, ఇమేజ్ జనరేషన్ అప్లికేషన్‌ను నిర్మించడానికి ఏమి అవసరం? మీరు క్రింది లైబ్రరీలను అవసరం:

- **python-dotenv**, మీ రహస్యాలను కోడ్ నుండి దూరంగా *.env* ఫైలులో ఉంచడానికి ఈ లైబ్రరీని అత్యంత సిఫార్సు చేయబడింది.
- **openai**, ఈ లైబ్రరీని OpenAI API తో ఇంటరాక్ట్ చేయడానికి ఉపయోగిస్తారు.
- **pillow**, Pythonలో చిత్రాలతో పనిచేయడానికి.

ఇప్పటివరకు చేయకపోతే, [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) పేజీలోని సూచనలను అనుసరించి Microsoft Foundry వనరును మరియు మోడల్‌ను సృష్టించండి. మోడల్‌గా **gpt-image-2**ను ఎంచుకోండి (తాజా Azure OpenAI చిత్రం మోడల్; DALL·E అనేది పాతది).

1. క్రింది కంటెంట్‌తో *.env* ఫైల్ను సృష్టించండి:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    ఈ సమాచారాన్ని మీ వనరు కోసం Microsoft Foundry పోర్టల్‌లో "Deployments" విభాగంలో కనుగొనండి.



1. పై లైబ్రరీలను *requirements.txt* అనే ఫైలులో ఇలా సేకరించండి:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. తర్వాత, వర్చువల్ ఎన్వైరమెంట్‌ను సృష్టించి లైబ్రరీలను ఇన్స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> విండోస్ కోసం, మీ వర్చువల్ ఎన్విరాన్మెంట్ సృష్టించడానికి మరియు యాక్టివేట్ చేయడానికి క్రింది కమాండ్లను ఉపయోగించండి:

    ```bash

    ````

1. *app.py* అనే ఫైల్లో క్రింది కోడ్‌ను జోడించండి:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv ని దిగుమతి చేయండి
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) క్లయింట్‌ను క్రమబద్ధీకరించండి.
    # చిత్రం నమూనాలు తాజా API వెర్షన్ అవసరం — మీ నమూనాకు కావలసినదాన్ని కనుగొనేందుకు Foundry డాక్యూమెంటేషన్‌ను తనిఖీ చేయండి.
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # చిత్రం జనరేషన్ API ఉపయోగించి ఒక చిత్రాన్ని సృష్టించండి
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # ఉదాహరణకు "gpt-image-2"
        )
        # నిల్వ చేస్తున్న చిత్రానికి డైరెక్టరీ సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')

        # డైరెక్టరీ లేనట్లయితే, దాన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # చిత్రం మార్గాన్ని ప్రారంభించండి (ఫైల్ రకం png కావాలి)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image నమూనాలు చిత్రాన్ని URL కాదు base64 (b64_json) గా ఇస్తాయి
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # చిత్రాన్ని డీఫాల్ట్ చిత్రం వీక్షకులో చూపించండి
        image = Image.open(image_path)
        image.show()

    # లోపాలను పట్టుకోండి
    except BadRequestError as err:
        print(err)

    ```

ఇక్కడ కోడ్‌ను వివరించుకుందాం:

- ముందుగా, మేము అవసరమైన లైబ్రరీలను దిగుమతి చేసుకుంటాము, ఇందులో OpenAI లైబ్రరీ, dotenv లైబ్రరీ, base64 మాడ్యూల్, మరియు Pillow లైబ్రరీ ఉన్నాయి.


- ఆ తర్వాత, మేము Azure OpenAI (Microsoft Foundry) క్లయింట్‌ను కాన్ఫిగర్ చేస్తాము.

    ```python

      api_key=os.environ['AZURE_OPENAI_API_KEY'],

    ```

- తరువాత, మేము చిత్రం రూపొందిస్తాము:

    ```python
    generation_response = client.images.generate(

    ```

    `gpt-image` మోడల్స్ చిత్రం **base64** స్ట్రింగ్‌గా `data[0].b64_json` లో ఇస్తాయి. మేము దీన్ని బైట్స్‌కి డీకోడ్ చేసి ఒక ఫైలుగా వ్రాస్తాము — డౌన్లోడ్ చేసే URL లే లేదు.

- చివరిగా, మేము చిత్రాన్ని తెరిచి స్టాండర్డ్ చిత్రం వీక్షకుని ఉపయోగించి చూపిస్తాము:

### చిత్రం రూపొందింపుపై మరిన్ని వివరాలు

`images.generate` యొక్క పారామితులను చూద్దాం:

- **prompt**, చిత్రం రూపొందించడానికి ఉపయోగించే టెక్స్ట్ ప్రాంప్ట్. ఇక్కడ ఇది "కుక్కపిల్లను గుర్రంపై, లాలిపాప్ పట్టుకొని, మెండుతున్న ఒక పొలంలో దాఫోడిల్స్ పెరుగుతున్నవి".
- **size**, రూపొందించిన చిత్ర పరిమాణం (ఉదాహరణకు `1024x1024`, `1536x1024`, `1024x1536` లేదా `"auto"`).
- **n**, రూపొందించబడిన చిత్రాల సంఖ్య. ఇక్కడ మేము ఒకదాన్ని రూపొందిస్తున్నాము.
- **model**, మీ చిత్రం మోడల్ డిప్లాయ్‌మెంట్ పేరు (ఉదాహరణకు `gpt-image-2`).

> చిత్రం మోడల్స్ `temperature` అనే పారామితిని తీసుకోవు — అది టెక్స్ట్-సృష్టి నియంత్రణ. వైవిధ్యానికి, APIని మళ్ళీ పిలవండి; వైవిధ్యం తగ్గించడానికి, మీ ప్రాంప్ట్ మరింత స్పష్టంగా చేయండి.

## చిత్రం రూపొందింపులో అదనపు సామర్థ్యాలు

మీరు కొద్ది ప్యాథాన్ కోడ్‌తో చిత్రం రూపొందించడం చూసారు. `gpt-image` మోడల్స్ ఇప్పటికే ఉన్న చిత్రాన్ని **మార్పు** చేయగలవు. మీరు ఒక చిత్రం, ఐచ్ఛిక **మాస్క్** (మార్పు చేయాల్సిన ప్రాంతాన్ని సూచిస్తుంది) మరియు ప్రాంప్ట్ అందిస్తే, చిత్రం భాగాన్ని మార్చవచ్చు — ఉదాహరణకు, మా కుక్కపిల్లకు టోపీ వేసుకోవచ్చు.

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# ఎడిట్లు కూడా base64 గా తిరిగి పంపబడతాయి
```
