ఈ ప్రక్రియతో, DALL-E, రూపొందించే చిత్రంలోని లక్షణాలు, వస్తువులు, స్వభావాలు మరియు మరిన్ని నియంత్రిస్తుంది. అయినప్పటికీ, DALL-E 2 మరియు 3 జనరేటెడ్ చిత్రంపై మరింత నియంత్రణ కలిగి ఉంటాయి,  


## మీ మొదటి చిత్రం సృష్టి అనువర్తనాన్ని నిర్మించడం

కాబట్టి చిత్రం సృష్టి అనువర్తనం నిర్మించడానికి ఏమి అవసరం? మీరు క్రింది లైబ్రరీలను అవసరం:

- **python-dotenv**, మీ రహస్యాలను కోడ్ నుండి దూరంగా *.env* ఫైల్లో ఉంచేందుకు ఈ లైబ్రరీని వినియోగించమని విడిగా సిఫార్సు చేయబడింది.
- **openai**, ఈ లైబ్రరీని మీరు OpenAI APIతో సంభాషించడానికి ఉపయోగిస్తారు.
- **pillow**, Pythonలో చిత్రాలతో పనిచేయడానికి.
- **requests**, HTTP అభ్యర్థనలను చేయడంలో మీకు సహాయం చేస్తుంది.

ఇంకా చేయని ఉంటే, [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) పేజీలోని సూచనలు అనుసరించి Azure OpenAI వనరు మరియు మోడల్ సృష్టించండి. మోడల్‌గా **gpt-image-1**ని ఎంచుకోండి (ప్రస్తుత తరం Azure OpenAI చిత్రం మోడల్; DALL-E 3 పాతది మరియు కొత్త పంపిణీలకు ఇక అందుబాటులో లేదు).

1. క్రింది అంశాలతో *.env* అనే ఫైల్ సృష్టించండి:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    ఈ సమాచారం మీరు సృష్టించిన వనరుకు Microsoft Foundry పోర్టల్ యొక్క "Deployments" విభాగంలో కనిపిస్తుంది.



1. పై లైబ్రరీలను *requirements.txt* అనే ఫైల్లో ఈ విధంగా సంగ్రహించండి:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. తరువాత, వర్చువల్ ఎన్విరాన్‌మెంట్ సృష్టించి లైబ్రరీలను ఇన్‌స్టాల్ చేయండి:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> విండోస్‌ కోసం, మీ వర్చువల్ ఎన్విరాన్‌మెంట్‌ని సృష్టించడానికి మరియు యాక్టివేట్ చేయడానికి కింది కమాండ్లను ఉపయోగించండి:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* అనే ఫైల్లో కింద చూపిన కోడ్‌ని జోడించండి:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # dotenv ను దిగుమతి చేసుకోండి
    dotenv.load_dotenv()

    # Azure OpenAI సేవ క్లయింట్‌ను కాన్ఫిగర్ చేయండి
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # ఇమేజ్ జనరేషన్ API ఉపయోగించి చిత్రం సృష్టించండి
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్‌ను ఇక్కడ నమోదు చేయండి
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # నిల్వచేయబడిన చిత్రానికి డైరెక్టరీని సెట్ చేయండి
        image_dir = os.path.join(os.curdir, 'images')
    
        # డైరెక్టరీ లేకపోతే, దీన్ని సృష్టించండి
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # చిత్రం మార్గాన్ని ప్రారంభించండి (ఫైల్ రకం png అవ్వాలి అని గమనించండి)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # సృష్టించబడిన చిత్రాన్ని పొందండి
        image_url = generation_response.data[0].url  # స్పందన నుండి చిత్రం URL తీసుకోండి
        generated_image = requests.get(image_url).content  # చిత్రాన్ని డౌన్లోడ్ చేయండి
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # డిఫాల్ట్ చిత్రం వీక్షణ యంత్రంలో చిత్రాన్ని ప్రదర్శించండి
        image = Image.open(image_path)
        image.show()
    
    # తప్పిదాలను పట్టు కుందండి
    except openai.BadRequestError as err:
        print(err)

    ```

- మొదటగా, మేము అవసరమైన లైబ్రరీలను, అందులో OpenAI లైబ్రరీ, dotenv లైబ్రరీ, requests లైబ్రరీ, మరియు Pillow లైబ్రరీని దిగుమతి చేసుకుంటాము.

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    ```

- తర్వాత, మేము *.env* ఫైల్ నుండి పరిసర వేరియబుల్స్‌ని లోడ్ చేస్తాము.

    ```python
    # డాట్ Envనుంచి దిగుమతి చేసుకోండి
    dotenv.load_dotenv()
    ```

- ఆ తర్వాత, మేము Azure OpenAI సర్వీస్ కస్టమర్‌ను కాన్ఫిగర్ చేస్తాము.

    ```python
    # Azure OpenAI సేవ క్లయింట్‌ను కాన్ఫిగర్ చేయండి
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )
    ```

- తరువాత, మేము చిత్రాన్ని సృష్టిస్తాము:

    ```python
    # ఇమేజ్ జనరేషన్ API ఉపయోగించి ఒక చిత్రం సృష్టించండి
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ఇక్కడ మీ ప్రాంప్ట్ టెక్స్ట్ నమోదు చేయండి
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    పై కోడ్ JSON ఆబ్జెక్ట్‌తో స్పందిస్తుంది, ఇందులో సృష్టించిన చిత్ర URL ఉంటుంది. ఆ URL‌ను ఉపయోగించి చిత్రాన్ని డౌన్‌లోడ్ చేసి ఫైల్‌గా సేవ్ చేయవచ్చు.

- చివరగా, మేము చిత్రాన్ని తెర చేసి సాధారణ చిత్రం వీక్షణ పరికరంతో ప్రదర్శిస్తాము:

    ```python
    image = Image.open(image_path)
    image.show()
    ```
    
### చిత్రాన్ని సృష్టించడంపై మరింత వివరాలు

చిత్రాన్ని సృష్టించే కోడ్‌ను మరింత వివరంగా చూద్దాం:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # మీ ప్రాంప్ట్ టెక్స్ట్ ఇక్కడ నమోదు చేయండి
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, చిత్రం సృష్టించడానికి ఉపయోగించే టెక్ట్స్ ప్రాంప్ట్. ఈ సందర్భంలో, మేము "కుందేలు గుర్రంపై, ఒక లాలిపాప్ పట్టుకుని, పువ్వులు పెరుగుతున్న మబ్బుమనుసుములో" అనే ప్రాంప్ట్ ఉపయోగిస్తున్నాము.
- **size**, సృష్టించబడిన చిత్ర పరిమాణం. ఈ సందర్భంలో, 1024x1024 పిక్సెల్స్ పరిమాణం ఉన్న చిత్రాన్ని సృష్టిస్తున్నాము.
- **n**, సృష్టించబడ్డ చిత్రాల సంఖ్య. ఈ సందర్భంలో, ఒక్క చిత్రాన్ని సృష్టిస్తున్నాము.

తరువాతి సెక్షన్‌లో మనం చిత్రాలతో చేయగల మరిన్ని అంశాలను చూడబోతున్నాము.

## చిత్రం సృష్టి యొక్క అదనపు సామర్థ్యాలు

ఇప్పటివరకు, కొన్ని పాఠాలతో మనం చిత్రం ఎలా సృష్టించగలమో చూశాము. అయినప్పటికీ, చిత్రాలతో మరిన్ని పనులు చేయవచ్చు.

మీరు ఈ క్రింది పనులు కూడా చేయవచ్చు:

- **సవరణలు చేయగలరు**. ఇప్పటికే ఉన్న చిత్రానికి మాస్క్ మరియు ప్రాంప్ట్ ఇవ్వడం ద్వారా, మీరు చిత్రాన్ని మార్చవచ్చు. ఉదాహరణకు, చిత్రంలో ఒక భాగానికి ఏదైనా జోడించవచ్చు. మన కుందేలు చిత్రాన్ని ఊహించండి, మీరు కుందేలుపై టోపీ జోడించవచ్చు. మీరు ఎలా చేసుకొనాలి అంటే, చిత్రం, మాస్క్ (మార్పు చేయవలసిన ప్రాంతాన్ని గుర్తించే) మరియు టెక్ట్స్ ప్రాంప్ట్ ఇవ్వాలి.

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    బేస్ చిత్రం కుందేలునే ఉంటే, తుది చిత్రంలో దానిపై టోపీ ఉంటుంది.
