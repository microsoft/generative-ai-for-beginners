# టెక్ట్స్ జనరేషన్ యాప్స్ నిర్మించండి

మీరు ఇప్పటి వరకు ఈ పాఠ్యক্রমంలో చూసిన విధంగా ప్రాంప్ట్స్ వంటి ముఖ్యమైన సూత్రాలు మరియు "ప్రాంప్ట్ ఇంజనీర్" అనే ఒక మొత్తం విభాగం కూడా ఉంది. మీరు సహచరంగా పనిచేసే అనేక సాధనాలు ఉన్నాయండి, ఉదాహరణకు ChatGPT, Office 365, Microsoft Power Platform, ఇవి ప్రాంప్ట్స్ ఉపయోగించి ఏదైనా పనిని పూర్తి చేయడంలో మీకు సహాయం చేస్తాయి.

మీ యాప్ లో ఇలాంటి అనుభవాన్ని జోడించేందుకు, మీరు ప్రాంప్ట్స్, కంప్లీషన్స్ వంటి సూత్రాలను అర్థం చేసుకోవాలి మరియు పనిచేసే లైబ్రరీ ని ఎంచుకోవాలి. ఇది ఈ అధ్యాయంలో మీరు నేర్చుకోబోయేది.

## ಪರಿಚయం

ఈ అధ్యాయంలో, మీరు:

- openai లైబ్రరీ మరియు దాని ప్రాథమిక సూత్రాలు గురించి నేర్చుకొండి.
- openai ఉపయోగించి ఒక టెక్ట్స్ జనరేషన్ యాప్ నిర్మించండి.
- ప్రాంప్ట్, టెంపరేచర్, టోకెన్స్ వంటి సూత్రాలను ఎలా ఉపయోగించాలో అర్థం చేసుకోండి.

## నేర్చుకునే లక్ష్యాలు

ఈ పాఠం పూర్తికాక, మీరు చేయగలరు:

- ఒక టెక్ట్స్ జనరేషన్ యాప్ అంటే ఏమిటో వివరించగలగడం.
- openai ఉపయోగించి ఒక టెక్ట్స్ జనరేషన్ యాప్ తయారుచేయడం.
- మీ యాప్ లో టోకెన్స్ సంఖ్యని, టెంపరేచర్‌ని సమంజసంగా మార్చడం ద్వారా వేరియబుల్ అవుట్పుట్ రూపొందించడం.

## టెక్ట్స్ జనరేషన్ యాప్ అంటే ఏమిటి?

సాధారణంగా మీరు యాప్ రూపొందించినప్పుడు, అది క్రింది విధంగా ఇన్‌టర్‌ఫేస్ ఉంటుంది:

- కమాండ్ ఆధారిత. కన్సోల్ యాప్స్ లాంటి కమాండ్ ఆధారిత యాప్స్ ఉంటాయి, మీరు కమాండ్ టైప్ చేస్తారు, అది ఒక పని చేస్తుంది. ఉదాహరణకు, `git` ఒక కమాండ్ ఆధారిత యాప్.
- యూజర్ ఇంటర్‌ఫేస్ (UI). కొంత యాప్స్ గ్రాఫికల్ యూజర్ ఇంటర్‌ఫేస్ (GUIs) కలిగివుంటాయి, మీరు బటన్లు క్లిక్ చేయవచ్చు, టెక్ట్స్ ఇన్‌పుట్ చేయవచ్చు, ఆప్షన్స్ ఎంచుకోవచ్చు.

### కన్సోల్ మరియు UI యాప్స్ పరిమితాల

మీరు టైప్ చేసిన కమాండ్ తో ఒక కమాండ్ ఆధారిత యాప్ తో పోల్చండి:

- **ఇది పరిమితమైనది**. మీరు ఏ కమాండ్ అయినా టైప్ చేయలేరు, మాత్రమే యాప్ మద్దతు ఇచ్చే కమాండ్లే ఉంటాయి.
- **భాషా నిర్దిష్టం**. కొన్ని యాప్స్ అనేక భాషల్ని మద్దతు ఇస్తాయి, కానీ మీరు మద్దతు పొందిన భాషను మాత్రమే ఉపయోగిస్తారు, భాషను చేర్చవచ్చు అయినప్పటికీ.

### టెక్ట్స్ జనరేషన్ యాప్స్ ల ప్రయోజనాలు

టెక్ట్స్ జనరేషన్ యాప్ ఎలా వేరుగా ఉంటుంది?

టెక్ట్స్ జనరేషన్ యాప్ లో, మీరు ఎక్కువ లవచ్ఛితత్వం కలిగి ఉంటారు, కమాండ్ల సెట్ లేదా ఒక నిర్దిష్ట ఇన్‌పుట్ భాషకు పరిమితం కాకుండా, సహజ భాషలో యాప్ తో సంభాషించవచ్చు. మరొక ప్రయోజనం ఏమంటే, మీరు ఇప్పటికే విస్తృత సమాచార శీఘ్రం లో శిక్షణ పొందిన డేటా స్రోతో సంభాషిస్తున్నారు, ఇది సాధారణ యాప్ డేటాబేస్ లో పరిమితమైనవే ఉంటాయి.

### టెక్ట్స్ జనరేషన్ యాప్ తో నేను ఏమి నిర్మించగలను?

మీరు అనేక అంశాలు తయారుచేయవచ్చు. ఉదాహరణకు:

- **ఒక చాట్‌బాట్**. మీ కంపెనీ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు జవాబులిచ్చే చాట్‌బాట్ ఒక మంచి ఎంపిక కావచ్చు.
- **సహాయకుడు**. LLMs ను టెక్ట్స్ సారాంశం, టెక్ట్స్ నుంచి అంతరాళ సమాచారములు పొందడం, రిజ్యూమ్ లాంటి టెక్ట్స్ తయారు చేయడం వంటి పనులలో బాగుంటాయి.
- **కోడ్ సహాయకుడు**. మీరు ఉపయోగించే భాషా మోడల్ ఆధారంగా, కోడ్ రచనలో సహాయం చేసే కోడ్ అసిస్టెంట్ ని మీరు తయారుచేయవచ్చు. ఉదాహరణకు, GitHub Copilot మరియు ChatGPT లాంటి ఉత్పత్తులను కోడ్ రాయడానికి ఉపయోగించవచ్చు.

## నేను ఎలా ప్రారంభించగలను?

మీరు LLM తో ఇంటిగ్రేట్ కావడానికి రెండు ప్రధాన విధానాలు ఉన్నాయి:

- API ఉపయోగించండి. మీ ప్రాంప్ట్ తో వెబ్ అభ్యర్థనలు రూపొందించి, టెక్ట్స్ తయారవుతుంది.
- లైబ్రరీ ఉపయోగించండి. లైబ్రరీలు API కాల్స్ ని సులభతరం చేసే నిర్మాణంగా ఉంటాయి.

## లైబ్రరీలు/SDKs

LLMs తో పనిచేసేందుకు కొద్దిమంది ప్రఖ్యాత లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్ కు కనెక్ట్ కావడం మరియు ప్రాంప్ట్స్ పంపడం సులభతరం చేస్తుంది.

తరువాత, उच्च స్థాయిలో పనిచేసే లైబ్రరీలు కూడా ఉన్నాయి:

- **Langchain**. Langchain ప్రఖ్యాతి పొందినది మరియు Python ని మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel Microsoft లైబ్రరీ, ఇది C#, Python మరియు Java భాషలను మద్దతు ఇస్తుంది.

## openai ఉపయోగించి మొదటి యాప్

మన మొదటి యాప్ ఎలా నిర్మించాలో, ఎలాంటి లైబ్రరీలు అవసరం, ఎంత అవసరమో తెలుసుకుందాం.

### openai ఇన్‌స్టాల్ చేయండి

  > [!NOTE] ఈ దశ కోడ్స్‌పేస్ లేదా డెవ్ కంటైనర్ లో నడుపిస్తున్నట్లయితే అవసరం లేదు


OpenAI లేదా Azure OpenAI తో ఇంటראక్ట్ చేయడానికి అనేక లైబ్రరీలు ఉన్నాయి. C#, Python, JavaScript, Java వంటి అనేక ప్రోగ్రామింగ్ భాషలు ఉపయోగించవచ్చు.  
మనం `openai` Python లైబ్రరీని ఎంచుకున్నాం, కాబట్టి `pip` తో ఇన్‌స్టాల్ చేయబోతున్నాము.

```bash
pip install openai
```

మీరు ఈ నోట్‌బుక్ ను కోడ్స్‌పేస్ లేదా డెవ్ కంటైనర్ లో నడుపుతున్నట్లయితే కాకపోతే, మీ మెషీనులో [Python](https://www.python.org/) ని ఇన్‌స్టాల్ చేయాలి.

### ఒక రిసోర్స్ సృష్టించండి

మీరు ఇప్పటికే చేయకపోతే, ఈ క్రింది దశలను చేయాలి:

- Azure <https://azure.microsoft.com/free/> లో అకౌంట్ సృష్టించండి.
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) లో Azure OpenAI రిసోర్స్ సృష్టించండి. [create a resource](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) గైడ్ చూడండి.


### API కీ మరియు ఎండ్‌పాయింట్ లొకేట్ చేయండి

ఈ దశలో, 'openai' లైబ్రరీకి ఏ API కీ ఉపయోగించాలో చెప్పాలి. మీ API కీ తెలుసుకోవడానికి, మీ Azure OpenAI రిసోర్స్ లో "Keys and Endpoint" విభాగానికి వెళ్లి "Key 1" 값을 కాపీ చేసుకోండి.

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ఇప్పుడు ఈ సమాచారాన్ని కాపీ చేసుకున్న తర్వాత, లైబ్రరీలను దీన్ని ఉపయోగించాలని అందించండి.

> [!NOTE]
> మీ కోడ్ నుండి API కీని వేరుగా ఉంచటం మంచిది. మీరు వాతావరణ వేరియబుల్స్ ఉపయోగించవచ్చు.
> - వాతావరణ వేరియబుల్ `AZURE_OPENAI_API_KEY` ని మీ API కీగా `.env` ఫైల్ లో సెట్ చేయండి. మీరు ఈ కోర్సు గత వ్యాయామాలు పూర్తిచేసి ఉంటే, ఇది ఇప్పటికే సెట్ అయి ఉంటుంది.


### Azure కాన్ఫిగరేషన్ సెట్ చేయండి

మీరు Azure OpenAI ఉపయోగిస్తుంటే, ఇక్కడ విధంగా కాన్ఫిగరేషన్ సెట్ చేయాలి. Responses API Azure OpenAI **v1 ఎండ్‌పాయింట్** నుండి సేవ చేస్తుంది, కాబట్టి `OpenAI` క్లయింట్ ని `<your-endpoint>/openai/v1/` వైపు చూపిస్తారు:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

పై కోడ్ లో, మనం క్రింది విషయాలు సెట్ చేస్తున్నాము:

- `api_key`, ఇది మీ Azure పోర్టల్ లో ఉన్న API కీ.
- `base_url`, ఇది మీ Azure OpenAI ఎండ్‌పాయింట్ గాను మరియు `/openai/v1/` జత చేయబడినది. ఈ ఎండ్‌పాయింట్ Azure పోర్టల్ లో API కీ పక్కన ఉంటుంది. v1 ఎండ్‌పాయింట్ వలన `api_version` ఇవ్వడం అవసరం ఉండదు.
- `deployment`, మీరు Foundry పోర్టల్లో సృష్టించిన మోడల్ డిప్లాయ్‌మెంట్ పేరు.

> [!NOTE]
> `os.environ` వాతావరణ వేరియబుల్స్ చదివే మ్యాపింగ్ లాగా ఉంటుంది. మీరు దీన్ని `AZURE_OPENAI_API_KEY` మరియు `AZURE_OPENAI_ENDPOINT` వంటివి చదివేందుకు ఉపయోగించవచ్చు.

## టెక్ట్స్ తయారు చేయండి

టెక్ట్స్ తయారు చేసే విధానం `responses.create` విధానాన్ని ఉపయోగించడం. ఉదాహరణ:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

పై కోడ్ లో, మేము ఒక రెస్పాన్స్ ఆబ్జెక్ట్ ను సృష్టించి, మోడల్ మరియు ప్రాంప్ట్ ను అందిస్తున్నాము. తరువాత తయారైన టెక్ట్స్ ని ప్రింట్ చేస్తున్నాము.

### చాట్ రెస్పాన్సెస్
