# టెక్స్ట్ జనరేషన్ యాప్‌లు నిర్మించండి

మీరు ఇప్పటివరకు ఈ అభ్యాస కార్యక్రమం ద్వారా చూసినట్లుగా, ప్రాంప్ట్‌లు వంటి కొ core  అడుగుబెట్టిన అంశాలు మరియు "ప్రాంప్ట్ ఇంజనీరింగ్" అని పిలవబడే పూర్తిగా ఒక శాస్త్ర విధానం కూడా ఉంది. ChatGPT, Office 365, Microsoft Power Platform వంటి అనేక టూల్స్ ప్రాంప్ట్‌లను ఉపయోగించి ఏదైనా సాధించడానికి మిమ్మల్ని సహాయపడతాయి.

మీరు ఇలాంటిచ్ అనుభవాన్ని ఒక యాప్‌లో చేర్చాలనుకుంటే, ప్రాంప్ట్‌లు, కంప్లీషన్‌లు వంటి విషయాలను అర్థం చేసుకోవలసి ఉంటుంది మరియు పనిచేయడానికి ఒక లైబ్రరీ ఎంచుకోవలసి ఉంటుంది. మీరు ఈ అధ్యాయంలో నేర్చుకొనే దేనే ఇది.

## పరిచయం

ఈ అధ్యాయంలో మీరు:

- openai లైబ్రరీ మరియు దాని మౌలిక అంశాల గురించి తెలుసుకుంటారు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ యాప్‌ని నిర్మిస్తారు.
- టెక్స్ట్ జనరేషన్ యాప్ నిర్మించడానికి ప్రాంప్ట్, టెంపరేచర్ మరియు టోకెన్స్ వంటి ఆలోచనలను ఎలా ఉపయోగించాలో అర్థం చేసుకుంటారు.

## నేర్చుకునే లక్ష్యాలు

ఈ పాఠం చివరికి, మీరు చేయగలుగుతారు:

- టెక్స్ట్ జనరేషన్ యాప్ అంటే ఏమిటి అని వివరిస్తారు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ యాప్‌ను నిర్మిస్తారు.
- మీ యాప్‌లో ఎక్కువ లేదా తక్కువ టోకెన్స్ ఉపయోగించి, అలాగే టెంపరేచర్ మారుస్తూ విభిన్న అవుట్పుట్ కోసం సెట్ చేస్తారు.

## టెక్స్ట్ జనరేషన్ యాప్ అంటే ఏమిటి?

సాధారణంగా మీరు యాప్ నిర్మించినప్పుడు, దానికీ క్రింద ఇచ్చినట్లుగా ఒక ఇంటర్‌ఫేస్ ఉంటుంది:

- కమాండ్ ఆధారిత. కన్సోల్ యాప్‌లు సాధారణంగా మీరు కమాండ్ టైప్ చేస్తారు మరియు అది టాస్క్‌ను పూర్తి చేస్తుంది. ఉదాహరణకు, `git` అనేది కమాండ్ ఆధారిత యాప్.
- వాడుకరి ఇంటర్‌ఫేస్ (UI). కొందరు యాప్‌లు గ్రాఫికల్ యూజర్ ఇంటర్‌ఫేస్ (GUIs) కలిగి ఉంటాయి, మీరు బటన్లను క్లిక్ చేయవచ్చు, టెక్స్ట్ ను ఇన్‌పుట్ చేయవచ్చు, ఎంపికలు ఎంచుకోవచ్చు మరియు మరింత చేయవచ్చు.

### కన్సోల్ మరియు UI యాప్‌లు పరిమితం

మీరు కమాండ్ ఆధారిత యాప్ where మీరు కమాండ్ టైప్ చేస్తారు: తో పోల్చండి:

- **ఇది పరిమితమైనది**. మీరు ఏదైనా కమాండ్ టైప్ చేయలేరు, యాప్ మద్దతు ఇచ్చినవే మాత్రమే.
- **భాషా నిర్ధిష్టం**. కొన్ని యాప్‌లు అనేక భాషలను మద్దతు ఇస్తాయి, కానీ డిఫాల్ట్‌గా యాప్ ఒక నిర్ధిష్ట భాష కోసం మాత్రమే నిర్మించబడింది, మీరు ఇంకొన్ని భాషల మద్దతు జోడించగలిగినా.

### టెక్స్ట్ జనరేషన్ యాప్‌ల లాభాలు

టెక్స్ట్ జనరేషన్ యాప్ ఎలా భిన్నమయ్యిందంటే?

టెక్స్ట్ జనరేషన్ యాప్‌లో మీకు ఎక్కువ స్వేచ్ఛ ఉంటుంది, మీరు కమాండ్ల లేదా నిర్ధిష్ట భాషల ఇన్‌పుట్లకు పరిమితం కాలేదు. మీరు సహజ భాషలో యాప్‌తో సంభాషించవచ్చు. మరో లాభం ఏంటంటే, మీరు ఇప్పటికే విస్తృత సమాచార సేకరణపై ట్రైనింగ్ ఇచ్చిన డేటా మూలంతో పరస్పరం చేస్తున్నారా. సంప్రదాయ యాప్‌లు డేటాబేస్‌లో ఉన్న విషయాలతో పరిమితం అయి ఉండొచ్చు.

### టెక్స్ట్ జనరేషన్ యాప్‌తో నేను ఏమి నిర్మించగలను?

మీరు ఎన్నో వస్తువులు నిర్మించవచ్చు. ఉదాహరణగా:

- **చాట్‌బాట్**. మీ సంస్థ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు సమాధానం చెప్పే చాట్‌బాట్ మంచి ఎంపిక.
- **సహాయకుడు**. LLMలు టెక్స్ట్ యొక్క సంక్షిప్తీకరణ, టెక్స్ట్ నుండి అర్థాలు తీసుకోవడం, రిజ్యూమ్ లాంటి టెక్స్ట్ సృష్టించడం వంటి పనుల్లో చాలా బాగున్నాయి.
- **కోడ్ సహాయకుడు**. మీరు ఉపయోగించే భాషా మోడల్‌ను ఆధారపడి, కోడ్ రాయడంలో సహాయం చేసే కోడ్ అసిస్టెంట్‌ను నిర్మించవచ్చు. ఉదాహరణకు, మీరు GitHub Copilot లేదా ChatGPT వంటి ఉత్పత్తులను ఉపయోగిస్తూ కోడ్ రాయడంలో సహాయపడవచ్చు.

## నేను ఎలా ప్రారంభించగలను?

సరే, మీరు సాధారణంగా LLMతో ఇంటిగ్రేట్ కావడానికి ఈ క్రింది రెండు విధానాలు ఉంటాయి:

- API ఉపయోగించండి. ఇక్కడ మీరు మీ ప్రాంప్ట్‌తో వెబ్ రిక్వెస్ట్‌లు నిర్మించి, జెనరేట్ అయిన టెక్స్ట్ ను తిరిగి పొందుతారు.
- లైబ్రరీ ఉపయోగించండి. లైబ్రరీలు API కాల్స్‌ను సమీకరించి వాటిని సులభంగా ఉపయోగించడానికి సహాయ పడతాయి.

## లైబ్రరీలు/SDKలు

LLMలతో పని చేయడానికి పక్కా పేరుప్రఖ్యాత కొన్ని లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్‌కు సులభంగా కనెక్ట్ కావడానికి మరియు ప్రాంప్ట్‌లు పంపడానికి తోడ్పడుతుంది.

తరువాత, కొంతమందికి ఉన్నతస్థాయిలో ఆపరేట్ అయ్యే లైబ్రరీలు కూడా ఉన్నాయి:

- **Langchain**. Langchain పేరుకొచ్చింది మరియు Python ను మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel అనేది Microsoft ద్వారానే తయారు చేసిన లైబ్రరీ, ఇది C#, Python, మరియు Java భాషలను మద్దతు ఇస్తుంది.

### 2. VS Code (లేదా మీ ఇష్టమైన IDE)

మీరు Microsoft Foundry ప్రాజెక్ట్ నుండి మీ ఇష్టమైన IDEలో పని చేయడానికి కావలసిన అన్ని సమాచారం ఉంది. ఈ ఉదాహరణ VS Code ని చూపిస్తుంది

* మీ Foundry ప్రాజెక్ట్‌లో చాట్ మోడల్ డిప్లాయ్ చేయండి. ఈ పాఠంలో టెంపరేచర్ ఉదాహరణల కోసం **Llama** మోడల్ వాడండి, ఉదా. `Llama-3.3-70B-Instruct`, ఎందుకంటే గణనాత్మక మోడల్స్ వంటి `gpt-5-mini` కు `temperature`/`top_p` సహాయం లేదు.
* ప్రాజెక్ట్ యొక్క **ఎండ్‌పాయింట్** మరియు **API కీ**ని ఓవerview పేజీ నుండి కాపీ చేసుకోండి
* వాటిని భద్రపరచడానికి పర్యావరణ మార్పిడులను సృష్టించండి: `AZURE_INFERENCE_ENDPOINT` మరియు `AZURE_INFERENCE_CREDENTIAL` - బాష్, పవర్‌షెల్ మరియు విండోస్ కమాండ్ ప్రాంప్ట్‌లో ఉదాహరణలు అందుబాటులో ఉన్నాయి
* డిపెండెన్సీలు ఇన్‌స్టాల్ చేయండి: ```pip install azure-ai-inference```
* ప్రాథమిక నమూనా కోడ్‌ను ఒక .py ఫైల్ లో కాపీ చేయండి

## వ్యాయామం - ఒక రేసిపి జనరేటర్ నిర్మించండి

ఇప్పుడు మనం ఒక సన్నివేశాన్ని ఆడిపోయినంతవరకు, ప్రదర్శించబడిన సన్నివేశానికి సరిపోయే కోడ్ రాయండి. అలా చేయడానికి, ఈ దశలను అనుసరించండి:

1. ఉన్న ఫైల్‌ను ప్రారంభ బిందువుగా ఉపయోగించండి
1. ఒక `prompt` వేరియబుల్ సృష్టించి కింది విధంగా నమూనా కోడ్‌ను మార్చండి:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

మీరు ఇప్పుడు కోడ్ ను నడిపితే, మీరు క్రిందలా ఫలితం చూడగలరు:

1. కింది విధంగా కోడ్‌ను మార్చుదాం:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
