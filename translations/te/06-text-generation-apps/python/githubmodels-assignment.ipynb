# వచనం స్రష్టి యాప్‌లు తయారు చేయండి

మీరు ఇప్పటివరకు ఈ పాఠ్యक्रमంలో చూడటమేనట్టి, ప్రాంప్ట్ల వంటి ముఖ్యమైన భావనలు మరియు "ప్రాంప్ట్ ఇంజనీరింగ్" అనే పూర్తి శాస్త్రం కూడా ఉందని తెలుసుకున్నున్నారు. మీరు ఇంతవరకు పరస్పర చర్య తీసుకునే అనేక సాధనాలు, వంటి ChatGPT, Office 365, Microsoft Power Platform మరియు ఇంకా ఎన్నో, మీరు ఏదో సాధించడానికి ప్రాంప్ట్‌లను ఉపయోగించడం ద్వారా మిమ్మల్ని సహాయపడతాయి.

మీరు అలాంటి అనుభవాన్ని యాప్‌లో చేర్చుకోవాలంటే, ప్రాంప్ట్‌లు, కాంప్లిషన్స్ లాంటివి వంటి భావనలను అర్థం చేసుకోవాలి మరియు ఒక లైబ్రరీని ఎంచుకొని పని చేయాలి. ఇదే మీరు ఈ అధ్యాయంలో నేర్చుకునేది.

## పరిచయం

ఈ అధ్యాయంలో, మీరు:

- openai లైబ్రరీ మరియు దాని ప్రధాన భావనలపై తెలుసుకోండి.
- openai ఉపయోగించి వచనం స్రష్టి యాప్‌ను నిర్మించండి.
- ప్రాంప్ట్, ఉష్ణోగ్రత, మరియు టోకెన్స్ వంటి భావనలను ఎలా ఉపయోగించటం ద్వారా వచనం స్రష్టి యాప్‌ను తయారు చేయాలో అర్థం చేసుకోండి.

## నేర్చుకొనే లక్ష్యాలు

ఈ పాఠం చివరికి, మీరు చేయగలుగుతారు:

- వచనం స్రష్టి యాప్ అంటే ఏమిటి అని వివరించగలగాలి.
- openai ఉపయోగించి వచనం స్రష్టి యాప్‌ను నిర్మించగలగాలి.
- మరిన్ని లేదా తక్కువ టోకెన్స్ ఉపయోగించి, అలాగే ఉష్ణోగ్రతను మార్చి, విభిన్నమైన ఫలితాన్ని పొందడానికి మీ యాప్‌ను కాన్ఫిగర్ చేయగలగాలి.

## వచనం స్రష్టి యాప్ అంటే ఏమిటి?

సాధారణంగా మీరు యాప్‌ను నిర్మించేటప్పుడు దానికో కొన్ని రకమైన ఇంటర్ఫేస్ ఉంటుంది, ఉదాహరణకు:

- ఆదేశాలతో నడిచే విధానం. కన్సోల్ యాప్‌లు సాధారణంగా మీరు ఆదేశం టైప్ చేస్తారు మరియు అది పని చేస్తుంది. ఉదాహరణకు, `git` ఆదేశాలతో నడిచే యాప్.
- వినియోగదారుడు ఇంటర్ఫేస్ (UI). కొన్ని యాప్‌లకు గ్రాఫికల్ యూజర్ ఇంటర్ఫేస్ (GUI) ఉంటుంది, అక్కడ మీరు బటన్లు నొక్కి, వచనం ఇన్పుట్ చేసి, ఎంపికలు చేసుకోవచ్చు.

### కన్సోల్ మరియు UI యాప్‌లు పరిమితమైనవి

ఆదేశాలతో నడిచే యాప్‌తో పోల్చు మీరు ఆదేశం టైప్ చేస్తారు:

- **ఇది పరిమితం**. మీరు ఏదైనా ఆదేశాన్ని టైప్ చేయలేరు, యాప్ మద్దతు ఇచ్చే ఆదేశాలకే పరిమితం.
- **భాషాపరమైనది**. కొన్ని యాప్‌లు అనేక భాషలను మద్దతు ఇస్తున్నా, సాదారణంగా యాప్ ఒక ప్రత్యేక భాష కోసం నిర్మించబడుతుంది, మరిన్ని భాషా మద్దతు చేర్చినా.

### వచనం స్రష్టి యాప్‌ల ప్రయోజనాలు

కాబట్టి వచనం స్రష్టి యాప్ ఎలా వేరుగా ఉంటుంది?

వచనం స్రష్టి యాప్‌లో మీకు మరింత స్వేచ్ఛ ఉంటుంది, మీరు ఆదేశాల సెట్ లేదా ఒక ప్రత్యేక ఇన్పుట్ భాషకు పరిమితులు ఉండరు. బదులుగా, మీరు సహజ భాషను ఉపయోగించి యాప్‌తో పరస్పరం చేసుకోవచ్చు. ఇంకొక లాభం ఏమంటే, మీరు ఇప్పటికే విస్తృత సమాచారం శిఖరం మీద శిక్షణ పొందిన డేటా మూలంతో పరస్పరం చేస్తున్నారని, పిల్లి యాప్ డేటాబేస్‌లో ఉన్నదానికి పరిమితం కావచ్చు.

### వచనం స్రష్టి యాప్ తో నేను ఏమి నిర్మించగలను?

మీరు ఎన్నో విషయాలను నిర్మించవచ్చు. ఉదాహరణకు:

- **ఒక చాట్బాట్**. మీ కంపెనీ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు సమాధానం ఇచ్చే చాట్బాట్ సరైన ఎంపిక.
- **సహాయకుడు**. LLMలు వచనం సారాంశం, వచనంతో అర్థాలు తీయడం, రెస్యూమేస్ వంటి వచనం ఉత్పత్తి చేయడం లాంటి పనులలో మంచిదిగా పనిచేస్తాయి.
- **కోడ్ సహాయకుడు**. మీరు ఉపయోగించే భాషా మోడల్ మీద ఆధారపడి, మీరు కోడ్ రాయడంలో సహాయపడే కోడ్ ఆధారిత సహాయకుడు తయారు చేసుకోవచ్చు. ఉదాహరణకు, GitHub Copilot వంటి ఉత్పత్తులు మరియు ChatGPT మీకు కోడ్ రాయడంలో సహాయం చేస్తాయి.

## నేను ఎలా ప్రారంభించగలను?

సరే, మీరు సాధారణంగా LLMతో ఇంటిగ్రేట్ కావడానికి ఈ రెండు మార్గాలు ఉంటాయి:

- API ఉపయోగించండి. ఇక్కడ మీరు మీ ప్రాంప్ట్ తో వెబ్ రిక్వెస్టులు నిర్మించి, సృష్టించిన వచనం తిరిగి పొందుతారు.
- లైబ్రరీ ఉపయోగించండి. లైబ్రరీలు API కాల్‌లను సులభంగా ఉపయోగించడానికి కప్పింగ్ చేయడంలో సహాయపడతాయి.

## లైబ్రరీలు/SDKలు

LLMలతో పని చేయడానికి కొన్ని మంచి ప్రసిద్ధ లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్‌తో కనెక్ట్ అవ్వడం మరియు ప్రాంప్ట్‌లు పంపడం సులభం చేస్తుంది.

తర్వాత ఉన్న లెవల్ పై పనిచేసే లైబ్రరీలు, ఉదాహరణకు:

- **Langchain**. Langchain ప్రసిద్ధి చెందింది మరియు Python మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel మైక్రోసాఫ్ట్ అందించిన లైబ్రరీ, C#, Python, మరియు Java భాషలను మద్దతు ఇస్తుంది.

## వ్యాయామం - ఒక వంటకాల రూపొందించే ప్రక్రియ నిర్మించండి

ఇప్పుడు మనం ఒక సందర్భాన్ని ఆడిన తరువాత, ప్రదర్శించిన సందర్భానికి సరిపోయే కో드를 రాయండి. ఇలా చేయడానికి, ఈ చర్యలను పాటించండి:

1. ఉన్న ఫైల్‌ను ప్రారంభ బిందువుగా ఉపయోగించండి
1. ఒక `prompt` మెరుగుదలని సృష్టించి, క్రింది విధంగా ఉదాహరణ కోడ్‌ను మార్చండి:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

అద్భుతం, వచ్చే సాధనను ఎలా మెరుగుపరచాలో చూసే అవకాశం ఉంది. మెరుగుపర్చేందుకు, కోడ్‌ను సౌలభ్యవంతంగా చేయాలి, కాబట్టి పదార్థాలు మరియు వంటకాల సంఖ్య మెరుగుపడు�దగినవి మరియు మార్చుకునే అవకాశం ఉండాలి.


1. క్రింది విధంగా కోడ్‌ను మార్చుదాం:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
