# టెక్స్ట్ జనరేషన్ అనువర్తనాలను నిర్మించండి

మీరు ఈ పాఠ్యాంశం ద్వారా ఇప్పటిదాకా చూసినట్టు, ప్రాంప్ట్‌లు వంటి మౌలిక సూత్రాలు ఉన్నాయి మరియు అది "ప్రాంప్ట్ ఇంజనీరింగ్" అనే పూర్తిగా ఒక విషయశాఖ కూడా ఉంది. మీరు పరస్పరం చేసుకునే చాలా సాధనాలు, ఉదాహరణకు ChatGPT, Office 365, Microsoft Power Platform మరియు మరికొన్ని, ఏదో పనిచేయడానికి ప్రాంప్ట్‌లను ఉపయోగించడం మద్దతు ఇస్తాయి.

మీరు ఇలాంటివి ఒక అనువర్తనంలో చేర్చాలంటే, ప్రాంప్ట్‌లు, పూర్తీయిన టెక్స్ట్ (completions) లాంటి సూత్రాలను అర్థం చేసుకొని, పనిచేయడానికి ఒక లైబ్రరీని ఎంచుకోవాలి. ఇది ఈ అధ్యాయంలో మీరు తెలుసుకోబోతున్న విషయం.

## పరిచయం

ఈ అధ్యాయంలో, మీరు:

- openai లైబ్రరీ మరియు దాని మౌలిక సూత్రాల గురించి తెలుసుకుంటారు.
- openai ఉపయోగించి ఒక టెక్స్ట్ జనరేషన్ అనువర్తనం నిర్మిస్తారు.
- టెక్స్ట్ జనరేషన్ అనువర్తనం నిర్మించడానికి ప్రాంప్ట్, ఉష్ణోగ్రత (temperature), టోకెన్లు వంటి సూత్రాలని ఎలా ఉపయోగించాలో అర్థం చేసుకుంటారు.

## నేర్చుకోవడంపై లక్ష్యాలు

ఈ పాఠం చివరిలో, మీరు చేయగలుగుతారు:

- టెక్స్ట్ జనరేషన్ అనువర్తనం అంటే ఏమిటో వివరించండి.
- openai ఉపయోగించి ఒక టెక్స్ట్ జనరేషన్ అనువర్తనం నిర్మించండి.
- మీ అనువర్తనాన్ని ఎక్కువ లేదా తక్కువ టోకెన్లు ఉపయోగించడానికి మరియు ఉష్ణోగ్రతను మార్చడానికి కాన్ఫిగర్ చేసి, వైవిధ్యమైన అవుట్పుట్ పొందండి.

## టెక్స్ట్ జనరేషన్ అనువర్తనం అంటే ఏమిటి?

సాధారణంగా మీరు ఒక అనువర్తనం నిర్మించినప్పుడు దానిలో కొన్ని రకాల ఇంటర్ఫేస్లు ఉంటాయి, ఉదాహరణగా ఈ క్రింది విధంగా:

- కమాండ్ ఆధారితం. కన్సోల్ అనువర్తనాలు సాధారణంగా మీరు కమాండ్ టైప్ చేసి ఆ పని చేయించుకునేవి. ఉదాహరణకి, `git` ఒక కమాండ్ ఆధారిత అనువర్తనం.
- యూజర్ ఇంటర్‌ఫేస్ (UI). కొన్ని అనువర్తనాల్లో గ్రాఫికల్ యూజర్ ఇంటర్‌ఫేస్ (GUI) ఉంటుంది, మీరు బటన్లను క్లిక్ చేయడం, టెక్స్ట్ ఇచ్చుకోవడం, ఎంపికల్ని ఎంచుకోవడం మొదలైనవి చేస్తారు.

### కన్సోల్ మరియు UI అనువర్తనాలకు పరిమితులు ఉన్నాయి

మీరు ఒక కమాండ్ ఆధారిత అనువర్తనంలో కమాండ్ టైప్ చేసే సందర్భాన్ని సరిపోల్చండి:

- **ఇది పరిమితం**. మీరు ఎలాంటి కమాండ్ ఇవ్వలేరు, కేవలం ఆ అనువర్తనం మద్దతు ఇచ్చే కమాండ్లతో మాత్రమే పని చేస్తుంది.
- **భాషాపరమైనది**. కొన్ని అనువర్తనాలు ఎన్నో భాషలను మద్దతు ఇస్తాయి, కానీ ప్రారంభానికి ఒక నిర్ధిష్ట భాష కోసం తయారు చేయబడి ఉంటాయి, మరింత భాషల మద్దతు ఇవ్వగలరని అనుకుంటే అయినా.

### టెక్స్ట్ జనరేషన్ అనువర్తనాల ప్రయోజనాలు

అప్పుడు టెక్స్ట్ జనరేషన్ అనువర్తనం ఎలా వేరుగా ఉంటుంది?

టెక్స్ట్ జనరేషన్ అనువర్తనంలో మీరు ఎక్కువ స్వేచ్ఛ కలిగి ఉంటారు, మీరు కమాండ్ల సెట్ లేదా నిర్ధిష్ట ఇన్‌పుట్ భాష పైన పరిమితులేము ఉండవు. మీరు సహజ భాషను ఉపయోగించి అనువర్తనంతో పరస్పరం చేయవచ్చు. మరో ప్రయోజనం ఏమిటంటే, మీరు ఇప్పటికే విస్తృత సమాచారం సేకరణతో శిక్షణ పొందిన డేటా మూలంతో పని చేస్తున్నారు, సాధారణంగా ఒక డేటాబేస్ పరిమితి కలిగిన అనువర్తనంతో పోలిస్తే.

### టెక్స్ట్ జనరేషన్ అనువర్తనంతో నేను ఏమి నిర్మించగలను?

మీరు అనేక వాటిని నిర్మించవచ్చు. ఉదాహరణకి:

- **చాట్‌బాట్**. కంపెనీ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు చాట్‌బాట్ సమాధానమిచ్చడం అనువైన దిగువకు సరిపోతుంది.
- **సహాయకుడు**. LLM లు టెక్స్ట్ సారాంశం, ఇన్‌సైట్స్ తీసుకోవడం, రిజూమేలా టెక్స్ట్ తయారు చేయడం వంటి పనులు చక్కగా చేస్తాయి.
- **కోడ్ సహాయకుడు**. మీరు ఉపయోగించే భాషా మోడల్‌పై ఆధారపడి, మీరు కోడ్ రాయడంలో సహాయపడే కోడ్ అసిస్టెంట్‌ను నిర్మించవచ్చు. ఉదాహరణకి, GitHub Copilot వంటి ఉత్పత్తి లేదా ChatGPT ని ఉపయోగించి కోడ్ రాయడంలో సహాయం పొందవచ్చు.

## నేను ఎలా మొదలు పెట్టగలను?

నిజానికి, మీరు ఒక LLM మోడల్‌తో ఇంటిగ్రేట్ అవ్వడానికి రెండు ప్రధాన మార్గాలను అనుసరించాలి:

- API ఉపయోగించడం. ఇక్కడ మీరు మీ ప్రాంప్ట్‌తో వెబ్ రిక్వెస్ట్‌లను నిర్మించి, జనరేట్ అయిన టెక్స్ట్ తిరిగి పొందుతారు.
- లైబ్రరీ ఉపయోగించడం. లైబ్రరీలు API పిలుపులను సంకలనం చేస్తాయి మరియు వాటిని ఉపయోగించడం సులభతరం చేస్తాయి.

## లైబ్రరీలు/SDKలు

LLM లతో పనిచేయడానికి కొన్ని ప్రసిద్ధ లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్‌కి సులభంగా కనెక్ట్ కావటానికి మరియు ప్రాంప్ట్‌లు పంపడానికి సహాయపడుతుంది.

తర్వాత, కాస్త అధిక స్థాయి ఆపరేషన్ చేసే లైబ్రరీలు కూడా ఉన్నాయి:

- **Langchain**. Langchain బాగా ప్రముఖం మరియు Python మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel మైక్రోసాఫ్ట్ నుండి లైబ్రరీ, ఇది C#, Python, మరియు Java భాషలను మద్దతు ఇస్తుంది.

## openai ఉపయోగించి మొదటి అనువర్తనం

మనము ఎలా ఒక మొదటి అనువర్తనం నిర్మించాలో, ఏ లైబ్రరీలు అవసరమో, ఎంతో అవసరం అవుతుందో చూద్దాం.

### openai ను ఇన్‌స్టాల్ చేయండి

  > [!NOTE] ఈ దశ Codespaces లేదా Devcontainer లో నడుపుతున్నప్పుడు అవసరం లేదు


OpenAI లేదా Azure OpenAI తో పరస్పరం చేయడానికి చాలా లైబ్రరీలు ఉన్నాయి. మీరు C#, Python, JavaScript, Java మరియు మరిన్ని ప్రోగ్రామింగ్ భాషలను కూడా ఉపయోగించవచ్చు.  
మేము `openai` Python లైబ్రరీని ఎంచుకున్నాము, కావున దాన్ని ఇన్స్టాల్ చేయడానికి `pip` ఉపయోగిస్తాము.

```bash
pip install openai
```

మీరు ఈ నోట్బుక్‌ను Codespaces లేదా Dev Container లో నడుపుతున్నట్లయితే కాకపోతే, మీ యంత్రంలో [Python](https://www.python.org/) ని కూడా ఇన్స్టాల్ చేయాలి.

### ఒక వనరును సృష్టించి మీ API కీని కనుగొనండి

మీరు ఇప్పటివరకు చేయకపోతే, ఈ కింద ఉన్న దశలను చేయాలి:

- OpenAIలో ఖాతాను సృష్టించండి <https://platform.openai.com/signup>.
- ఇప్పుడు, మీ API కీలు పొందండి <https://platform.openai.com/api-keys>. 

>[!NOTE]
> మీరు మీ API కీని మీ కోడ్స్‌ల నుండి వేరువేరు ఉంచటం మంచిది. మీరు దీనిని ఎండురాట్మెంట్ వేరియబుల్స్ ద్వారా చేయవచ్చు.
> - `.env` ఫైల్‌లో `OPENAI_KEY` అనే వాతావరణ వేరియబుల్ గా మీ API కీని సెట్ చేయండి. మీరు ఇప్పటికే ఈ కోర్సు యొక్క పూర్వపు వ్యాయామాలను పూర్తి చేసినట్లయితే, సెట్ అయిపోయింది.
> - API కీ ఒకసారి మాత్రమే యాక్సెస్ అవుతుంది. అందువల్ల అది సరిగ్గా కాపీ అయిందని నిర్ధారించాలి. పనిచేయకపోతే, కీని తొలగించి కొత్తది జనరేట్ చేయడం మంచిది.


### OpenAI కాన్ఫిగరేషన్ సెట్ అప్

మీరు OpenAI ఉపయోగిస్తే, ఇలా కాన్ఫిగర్ చేయవచ్చు:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

పై భాగంలో మనం ఈ క్రింది విషయాలను సెట్ చేస్తున్నాము:

- `api_key`, ఇది OpenAI డ్యాష్బోర్డ్‌లో కనిపించే మీ API కీ.
- `deployment`, ఇది మీ GPT వెర్షన్.

> [!NOTE]
> `os.environ` అనేది వాతావరణ వేరియబుల్స్‌ని చదవడానికి ఉపయోగించే ఫంక్షన్. దీన్ని `OPENAI_API_KEY` లాగానే వేరియబుల్స్‌ని చదవడానికి ఉపయోగించచ్చు.

## టెక్స్ట్ ను ఉత్పత్తి చేయండి

టెక్స్ట్ ఉత్పత్తి చేయడానికి `responses.create` పద్ధతిని ఉపయోగిస్తారు. ఇక్కడ ఒక ఉదాహరణ చూపబడింది:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

పై కోడ్‌లో, మనం ఒక response ఆబ్జెక్ట్ సృష్టించి, ఉపయోగించాలనుకున్న మోడల్ మరియు ప్రాంప్ట్‌ను అందించాం. ఆ తరువాత ఉత్పత్తి అయిన టెక్స్ట్‌ను ప్రింట్ చేసాం.

### చాట్ స్పందనలు

ఇప్పటివరకు, మీరు ఒకే ప్రాంప్ట్ నుండి ఎలా టెక్స్ట్ ఉత్పత్తి చేయాలో చూశారు. Responses API చాట్‌ బాట్ల కోసం కూడా బాగా సరిపోతుంది - మీరు సంభాషణను నిర్మించడానికి `input`గా సందేశాల జాబితా ఇస్తారు. దీని ఉపయోగం ఇక్కడ ఉంది:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

print(response.output_text)
```

ఇప్పుడు మనకు వంటకాలను తయారు చేయగల పనిచేసే యాప్ ఉంది మరియు ఇది వినియోగదారుని నుండి ఇన్పుట్లపై ఆధారపడుతూ సడలింపుగా ఉంటుంది, వంటకాల సంఖ్య మాత్రమే కాదు, ఉపయోగించిన పదార్థాలు కూడా.

మరింత మెరుగుపరచడానికి, మనం క్రింది వాటిని జోడించాలని కోరుకుంటున్నాము:

- **పదార్థాలను ఫిల్టర్ చేయడం**. మనకు ఇష్టంలేని లేదా అలర్జీ ఉన్న పదార్థాలను ఫిల్టర్ చేయడం సామర్థ్యం కావాలి. ఈ మార్పు సాధించడానికి, మనం వున్న ప్రాంప్ట్‌ని సవరించి చివరగా ఫిల్టర్ షరతును ఇలా జోడించవచ్చు:


- **షాపింగ్ లిస్ట్ తయారు చేయడం**. మనకు ఇళ్లలో ఇప్పటికే ఉన్న వస్తువులను పరిగణలోకి తీసుకుని షాపింగ్ లిస్ట్ తయారుచేయాలి.

    ఈ ఫంక్షనాలిటీ కోసం, మనం ఒక్క ప్రాంప్ట్‌లోనే అన్నింటిని పరిష్కరించవచ్చు లేదా రెండు ప్రాంప్ట్‌లలో విడగొట్టవచ్చు. మనం రెండవ పద్దతిని ప్రయత్నిద్దాం. ఇక్కడ మేము అదనపు ప్రాంప్ట్‌ని జోడించమని సూచిస్తున్నాము, కానీ అది పనిచేయడానికి, మొన్నటి ప్రాంప్ట్ ఫలితాన్ని ప్రస్తుత ప్రాంప్ట్‌కి సందర్భంగా జోడించాల్సి ఉంటుంది.

    మొదటి ప్రాంప్ట్ ఫలితాన్ని ప్రింట్ చేసే భాగాన్ని కనుగొని, క్రింద ఇచ్చిన కోడ్‌ని అతని కింద జోడించండి:
