# టెక్స్ట్ జనరేషన్ అప్లికేషన్లు నిర్మించండి

ఈ సిలాబస్‌లో ఇప్పటి వరకు మీరు చూసినట్లుగా, ప్రాంప్ట్‌లు మరియు "ప్రాంప్ట్ ఇంజనీరింగ్" అనే పాఠశాఖ వంటి ప్రాథమిక భావనలు ఉన్నాయి. మీరు ఇంటరాక్ట్ చేయగల అనేక సాధనాలు, ఉదాహరణకు ChatGPT, Office 365, Microsoft Power Platform మరియు మరిన్ని, మీరు ఏదైనా సాధించేందుకు ప్రాంప్ట్‌లను ఉపయోగించడంలో మద్దతు ఇస్తాయి.

మీరు ఒక అప్లికేషన్‌కు ఇలాంటి అనుభవాన్ని జోడించాలనుకుంటే, ప్రాంప్ట్‌లు, కంప్లిషన్స్ వంటి భావనలను అర్థం చేసుకోవడం మరియు పని చేసుకునే లైబ్రరీని ఎంచుకోవడం అవసరం. ఈ అధ్యాయంలో మీరు ఇదే నేర్చుకోబోతున్నారు.

## పరిచయం

ఈ అధ్యాయంలో మీరు:

- openai లైబ్రరీ మరియు దాని ప్రాథమిక భావనలను తెలుసుకుంటారు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ అప్లికేషన్‌ను నిర్మిస్తారు.
- ప్రాంప్ట్, టెంపరేచర్, టోకెన్లు వంటి భావనలను ఉపయోగించి టెక్స్ట్ జనరేషన్ అప్లికేషన్‌ను ఎలా నిర్మించాలో అర్థం చేసుకుంటారు.

## అభ్యసన లక్ష్యాలు

ఈ పాఠ్యం ముగింపులో మీరు:

- టెక్స్ట్ జనరేషన్ అప్లికేషన్ అంటే ఏమిటి అని వివరించగలుగుతారు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ అప్లికేషన్‌ను నిర్మించగలుగుతారు.
- మీ అప్లికేషన్‌ను మరింత లేదా తక్కువ టోకెన్లు ఉపయోగించడానికి మరియు అద్భుతమైన అవుట్పుట్ కోసం టెంపరేచర్‌ను మార్చడానికి పరిమితి చేయగలుగుతారు.

## టెక్స్ట్ జనరేషన్ అప్లికేషన్ అంటే ఏమిటి?

సాధారణంగా మీరు ఒక అప్లికేషన్‌ను నిర్మించినప్పుడు దానికి కింది రకం ఇంటర్ఫేస్ ఉంటుంది:

- కమాండ్ ఆధారితంగా. కంసోల్ అప్లికేషన్లు సాధారణంగా మీరు ఒక కమాండ్ టైప్ చేస్తారు మరియు అది ఒక పనిని చేస్తుంది. ఉదాహరణకి, `git` ఒక కమాండ్ ఆధారిత అప్లికేషన్.
- యూజర్ ఇంటర్ఫేస్ (UI). కొన్ని అప్లికేషన్లు గ్రాఫికల్ యూజర్ ఇంటర్ఫేస్ (GUI) కలిగి ఉంటాయి, మీరు బటన్లు క్లిక్ చేస్తారు, టెక్స్ట్ ఇన్పుట్ చేస్తారు, ఎంపికలు ఎంచుకుంటారు మరియు మరిన్ని చేస్తారు.

### కంసోల్ మరియు UI అప్లికేషన్లు పరిమితంగా ఉంటాయి

మీరు ఒక కమాండ్ టైప్ చేసే కమాండ్ ఆధారిత అప్లికేషన్‌తో పోల్చండి: 

- **ఇది పరిమితం**. మీరు ఎలాంటి కమాండ్ కేవలం టైప్ చేయలేరు, కేవలం అప్లికేషన్ మద్దతు ఇచ్చే కమాండ్లను మాత్రమే.
- **భాషకు నిర్దిష్టం**. కొన్ని అప్లికేషన్లు అనేక భాషలను మద్దతు ఇస్తాయి, కానీ ప్రాథమికంగా అప్లికేషన్ ఒక నిర్దిష్ట భాష కోసం తయారు చేయబడింది, మీరు మరిన్ని భాషలను జోడించగలిగినా కూడా.

### టెక్స్ట్ జనరేషన్ అప్లికేషన్ల లాభాలు

అప్పుడు టెక్స్ట్ జనరేషన్ అప్లికేషన్ ఎలా వేరుగా ఉంటుంది?

టెక్స్ట్ జనరేషన్ అప్లికేషన్‌లో, మీరు మరింత సౌలభ్యంతో ఉంటారు, మీరు ఒక సెట్ కమాండ్లు లేదా ఒక నిర్దిష్ట ఇన్పుట్ భాషకు పరిమితం కాదు. మీరు సహజ భాషను ఉపయోగించి అప్లికేషన్‌తో ఇంటరాక్ట్ చేయవచ్చు. మరో లాభం ఏమిటంటే, మీరు ఇప్పటికే ఒక పెద్ద సమాచార నిధిపై శిక్షణ పొందిన డేటా స్రోత్నంతో ఇంటరాక్ట్ చేస్తున్నారు, ఎక్కడైతే సాంప్రదాయ అప్లికేషన్ డేటాబేస్‌లో ఉన్నదానితో పరిమితం కావచ్చు.

### టెక్స్ట్ జనరేషన్ అప్లికేషన్‌తో నేను ఏమి నిర్మించగలను?

మీరు అనేక రకాల విషయాలను నిర్మించవచ్చు. ఉదాహరణకు:

- **ఛాట్బాట్**. ఒక ఛాట్‌బాట్ మీ కంపెనీ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు సమాధానాలు ఇవ్వడం ఒక మంచి మేళవింపు కావచ్చు.
- **సహాయకుడు**. LLMs టెక్స్ట్‌ను సారాంశం చేయడం, టెక్స్ట్ నుండి పరిశీలనలు పొందడం, రిజ్యూమెస్ వంటి టెక్స్ట్ ఉత్పత్తి చేయడం వంటి పనులలో గొప్పగా ఉంటాయి.
- **కోడ్ అసిస్టెంట్**. మీరు ఉపయోగించే భాషా మోడల్‌పై ఆధారపడి, మీరు కోడ్ రాసుకోవడంలో సహాయపడే కోడ్ అసిస్టెంట్‌ను నిర్మించవచ్చు. ఉదాహరణకి, GitHub Copilot వంటి ఉత్పత్తులు మరియు ChatGPT కోడ్ రాయడంలో మీకు సహాయపడతాయి.

## నేను ఎలా ప్రారంభించగలను?

బాగుంటుంది, మీరు సాధారణంగా LLMతో ఇంటిగ్రేట్ అయ్యేందుకు ఈ రెండు విధానాలు ఉంటాయి:

- API ఉపయోగించండి. ఇక్కడ మీరు మీ ప్రాంప్ట్‌తో వెబ్ అభ్యర్థనలను నిర్మించి, ఉత్పన్నమైన పాఠ్యాన్ని తిరిగి పొందుతారు.
- లైబ్రరీ ఉపయోగించండి. లైబ్రరీలు API కాల్స్‌ను సంయుక్తం చేసి వాటిని ఉపయోగించుకోవటం సులభం చేస్తాయి.

## లైబ్రరీలు/SDKలు

LLMsతో పని చేయడానికి కొన్ని పాపులర్ లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్‌తో సులభంగా కనెక్ట్ అయ్యి ప్రాంప్ట్‌లను పంపడానికి సహాయపడుతుంది.

తరువాత ఉన్నవి ఉన్నత స్థాయి లైబ్రరీలు:

- **Langchain**. Langchain మంచి పేరు பெற்றది మరియు Python మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel Microsoft యొక్క లైబ్రరీ, ఇది C#, Python, మరియు Java భాషలను మద్దతు ఇస్తుంది.

## openai ఉపయోగించి మొదటి అప్లికేషన్

మనమెలా మొదటి అప్లికేషన్‌ని నిర్మించాలో, ఏ లైబ్రరీలు అవసరమో, ఎంత అవసరమో చూస్తాం.

### openai ఇన్‌స్టాల్ చేయండి

  > [!NOTE] ఈ దశ కోడ్స్పేస్‌లలో లేదా డెవ్ కంటైనర్‌లో నడిపించినట్లయితే అవసరం లేదు


OpenAI లేదా Azure OpenAIతో ఇంటరాక్ట్ చేయడానికి అనేక లైబ్రరీలు ఉన్నాయి. C#, Python, JavaScript, Java మరియు మరిన్ని వంటి అనేక ప్రోగ్రామింగ్ భాషలను ఉపయోగించుకోవచ్చు.  
మేము `openai` Python లైబ్రరీని ఎంచుకున్నాము, కాబట్టి మనం `pip` ద్వారా దాన్ని ఇన్‌స్టాల్ చేసుకుంటాము.

```bash
pip install openai
```

మీరు కోడ్స్పేస్‌లు లేదా డెవ్ కంటైనర్‌లో ఈ నోట్‌బుక్ నడుపుతుంటే కాకుండా, మీరు మీ యంత్రంలో [Python](https://www.python.org/)ని కూడా ఇన్‌స్టాల్ చేయాలి.

### రిసోర్స్ సృష్టించండి మరియు మీ API కీని కనుగొనండి

మీరు ఇంకా చేయకపోతే, ఈ క్రింది దశలను అనుసరించండి:

- OpenAI <https://platform.openai.com/signup>లో అకౌంట్ సృష్టించండి.
- ఇప్పుడు మీ API కీలు <https://platform.openai.com/api-keys> పొందండి.

>[!NOTE]
> మీ API కీని మీ కోడ్ నుండి వేరుగా ఉంచటం మంచిది. మీరు దీనిని ఎన్విరాన్‌మెంట్ వేరియబుల్స్ ద్వారా చేయవచ్చు.
> - ఎన్విరాన్‌మెంట్ వేరియబుల్ `OPENAI_KEY` ని మీ API కీతో .env ఫైల్‌లో సెట్ చేయండి. మీరు ఈ కోर्स్ యొక్క మునుపటి వ్యాయామాలను పూర్తి చేసినట్లయితే, అందరూ సెట్ అయ్యారు.
> - API కీ ఒక్కసారి మాత్రమే యాక్సెస్ చేయగలగడం ముఖ్యమైనది. కాబట్టి అది సరిగ్గా కాపీ అయ్యిందో లేదో చెక్ చేసుకోవడం అవసరం. పనిచేయకపోతే కీని తొలగించి కొత్తది సృష్టించాలి.


### OpenAI కాన్ఫిగరేషన్ సెటప్ చేయండి

మీరు OpenAI ఉపయోగిస్తే, కాన్ఫిగరేషన్ ఇలా సెటప్ చేయండి:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

పై కోడ్‌లో మేము క్రింది వాటిని సెటప్ చేయడం:

- `api_key`, ఇది OpenAI డ్యాష్‌బోర్డ్‌లో కనిపించే మీ API కీ.
- `deployment`, ఇది మీ GPT వెర్షన్.

> [!NOTE]
> `os.environ` అనేది ఎన్విరాన్‌మెంట్ వేరియబుల్స్ చదువుకునే ఫంక్షన్. మీరు దీన్ని `OPENAI_API_KEY` వంటి వేరియబుల్స్ చదవడానికి ఉపయోగించవచ్చు.

## టెక్స్ట్ ఉత్పత్తి చేయండి

టెక్స్ట్ ఉత్పత్తి చేయడానికి `responses.create` మెథడ్‌ను ఉపయోగిస్తారు. ఉదాహరణ ఇక్కడ ఉంది:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

పై కోడ్‌లో, మేము ఒక రిస్పాన్స్ ఒబ్జెక్ట్‌ను సృష్టించి, ఉపయోగించాల్సిన మోడల్ మరియు ప్రాంప్ట్ పంపిస్తాము. తరువాత మేము ఉత్పత్తి చేసిన టెక్స్ట్‌ను ప్రింట్ చేస్తాము.

### చాట్ ప్రతిస్పందనలు

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

మనకు ఇప్పుడు రెసిపీలు తయారుచేసే పనిచేయగల ఒక యాప్ ఉందీ, ఇది వినియోగదారుడి నుండి వచ్చిన ఇన్పుట్‌లపై ఆధారపడుతూ అనుకూలంగా ఉంటుంది, రెసిపీల సంఖ్య మరియు ఉపయోగించిన పదార్థాలపై కూడా.

దీన్ని మరింత మెరుగుపర్చేందుకు, క్రింది విషయాలను జోడించాలని అనుకుంటున్నాం:

- **పదార్థాలను ఫిల్టర్ చేయడం**. మనం ఇష్టపడని లేదా అలెర్జీ ఉన్న పదార్థాలను ఫిల్టర్ చేయగలగాలి. ఈ మార్పును చేయడానికి, మన ప్రాంప్ట్‌ను సవరిస్తూ చివరలో ఫిల్టర్ షరతును జోడించవచ్చు ఇలా:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    పైవైపు, మేము ప్రాంప్ట్ చివర `{filter}` ను జోడిస్తున్నాం మరియు వినియోగదారుని నుండి ఫిల్టర్ విలువను కూడా దోచుకుంటున్నాం.

    ప్రోగ్రామ్ నడిపేటప్పుడు ఒక ఉదాహరణ ఇన్పుట్ ఇప్పుడే ఇలానే ఉండొచ్చు:
    

    మీరు చూస్తున్నట్లే, పాలు కలిగిన ఏవైనా రెసిపీలు ఫిల్టర్ చేయబడ్డాయి. కానీ మీరు లాక్టోజ్ అసహనం ఉంటే, చీజ్ ఉన్న రెసిపీలను కూడా ఫిల్టర్ చేయాలనుకుంటే, అందుకు స్పష్టంగా ఉండాల్సిన అవసరం ఉంది.

- **షాపింగ్ లిస్ట్ తయారీ**. మనం ఇప్పటికే ఇంట్లో ఉన్న పదార్థాలను పరిగణనలోకి తీసుకుని షాపింగ్ లిస్ట్ తయారుచేయదలిచాం.

    ఈ కార్యాచరణ కోసం, అన్ని విషయాలను ఒకే ప్రాంప్ట్‌లో పరిష్కరించడానికి ప్రయత్నించవచ్చు లేదా రెండుగా విభజించవచ్చు. రెండవ విధానం ప్రయత్నిద్దాం. ఇక్కడ మనం అదనపు ప్రాంప్ట్ ని జోడించాలి అని సూచిస్తున్నాం, కానీ అది పనిచేయాలంటే, మన మొట్టమొదటి ప్రాంప్ట్ ఫలితాన్ని రెండవ ప్రాంప్ట్ కి సందర్భంగా జోడించాలి.

    కోడ్‌లో మొదటి ప్రాంప్ట్ నుండి ఫలితాన్ని ప్రింట్ చేస్తున్న భాగాన్ని కనుగొని, క్రింది కోడ్ కింద జోడించండి:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # ప్రతిస్పందనను ముద్రించండి

    ```
