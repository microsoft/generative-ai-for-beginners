# టెక్స్ట్ జనరేషన్ యాప్స్ నిర్మించండి

ఈ పాఠ్యक्रमం ద్వారా మీరు ఇప్పటివరకు చూసినట్లుగా ప్రాంప్ట్స్ వంటి ప్రాథమిక భావనలు మరియు "ప్రాంప్ట్ ఇంజనీరింగ్" అనే ఒక పూర్తి శాస్త్రం కూడా ఉంది. మీరు ఇంటరాక్ట్ చేయగల అనేక టూల్స్, ఉదాహరణకు ChatGPT, Office 365, Microsoft Power Platform మరియు మరిన్ని, మీరు ఏదైనా సాధించడానికి ప్రాంప్ట్స్ ఉపయోగించడాన్ని మద్దతు ఇస్తాయి.

మీరు అలాంటి అనుభవాన్ని ఒక యాప్‌లో చేర్చాలంటే, ప్రాంప్ట్స్, కంప్లీషన్స్ వంటి భావనలను అర్థం చేసుకోవాలి మరియు పని చేయడానికి ఒక లైబ్రరీని ఎంచుకోవాలి. ఇదే మీరు ఈ అధ్యాయంలో నేర్చుకోబోతున్నది.

## పరిచయం

ఈ అధ్యాయంలో, మీరు:

- openai లైబ్రరీ మరియు దాని ప్రాథమిక భావనలను నేర్చుకుంటారు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ యాప్ నిర్మిస్తారు.
- ప్రాంప్ట్, టెంపరేచర్, టోకెన్స్ వంటి భావనలను ఉపయోగించి టెక్స్ట్ జనరేషన్ యాప్ ఎలా నిర్మించాలో అర్థం చేసుకుంటారు.

## నేర్చుకునే లక్ష్యాలు

ఈ పాఠం చివరికి, మీరు చేయగలుగుతారు:

- టెక్స్ట్ జనరేషన్ యాప్ అంటే ఏమిటో వివరించగలరు.
- openai ఉపయోగించి టెక్స్ట్ జనరేషన్ యాప్ నిర్మించగలరు.
- మీ యాప్‌ను ఎక్కువ లేదా తక్కువ టోకెన్స్ ఉపయోగించడానికి మరియు టెంపరేచర్ మార్చడానికి కాన్ఫిగర్ చేయగలరు, తద్వారా విభిన్న అవుట్పుట్ పొందవచ్చు.

## టెక్స్ట్ జనరేషన్ యాప్ అంటే ఏమిటి?

సాధారణంగా మీరు ఒక యాప్ నిర్మించినప్పుడు దానికి క్రింది విధమైన ఇంటర్‌ఫేస్ ఉంటుంది:

- కమాండ్ ఆధారిత. కన్సోల్ యాప్స్ సాధారణంగా మీరు ఒక కమాండ్ టైప్ చేసి ఒక పని చేస్తాయి. ఉదాహరణకు, `git` ఒక కమాండ్ ఆధారిత యాప్.
- యూజర్ ఇంటర్‌ఫేస్ (UI). కొన్ని యాప్స్ గ్రాఫికల్ యూజర్ ఇంటర్‌ఫేస్ (GUIs) కలిగి ఉంటాయి, అక్కడ మీరు బటన్లు క్లిక్ చేయడం, టెక్స్ట్ ఇన్‌పుట్ చేయడం, ఎంపికలు ఎంచుకోవడం మొదలైనవి చేస్తారు.

### కన్సోల్ మరియు UI యాప్స్ పరిమితులు

మీరు ఒక కమాండ్ టైప్ చేసే కమాండ్ ఆధారిత యాప్‌తో పోల్చండి:

- **ఇది పరిమితమైనది**. మీరు ఏదైనా కమాండ్ టైప్ చేయలేరు, యాప్ మద్దతు ఇచ్చే కమాండ్లకే పరిమితం.
- **భాషా నిర్దిష్టం**. కొన్ని యాప్స్ అనేక భాషలను మద్దతు ఇస్తాయి, కానీ డిఫాల్ట్‌గా యాప్ ఒక నిర్దిష్ట భాష కోసం నిర్మించబడుతుంది, మీరు మరిన్ని భాషా మద్దతు చేర్చినా.

### టెక్స్ట్ జనరేషన్ యాప్స్ లాభాలు

అయితే టెక్స్ట్ జనరేషన్ యాప్ ఎలా వేరుగా ఉంటుంది?

టెక్స్ట్ జనరేషన్ యాప్‌లో మీరు ఎక్కువ స్వేచ్ఛ కలిగి ఉంటారు, మీరు ఒక నిర్దిష్ట కమాండ్ల సెట్ లేదా ఒక నిర్దిష్ట ఇన్‌పుట్ భాషకు పరిమితం కాదు. బదులుగా, మీరు సహజ భాషను ఉపయోగించి యాప్‌తో ఇంటరాక్ట్ చేయవచ్చు. మరో లాభం ఏమిటంటే, మీరు ఇప్పటికే విస్తృత సమాచార సేకరణపై శిక్షణ పొందిన డేటా మూలంతో ఇంటరాక్ట్ చేస్తున్నారు, అయితే సాంప్రదాయ యాప్ డేటాబేస్‌లో ఉన్నదానితో పరిమితం కావచ్చు.

### టెక్స్ట్ జనరేషన్ యాప్‌తో నేను ఏమి నిర్మించగలను?

మీరు అనేక విషయాలు నిర్మించవచ్చు. ఉదాహరణకు:

- **చాట్‌బాట్**. మీ కంపెనీ మరియు దాని ఉత్పత్తుల గురించి ప్రశ్నలకు సమాధానం ఇచ్చే చాట్‌బాట్ మంచి ఎంపిక కావచ్చు.
- **సహాయకుడు**. LLMలు టెక్స్ట్ సారాంశం, టెక్స్ట్ నుండి అవగాహన పొందడం, రిజ్యూమ్‌ల వంటి టెక్స్ట్ ఉత్పత్తి చేయడం వంటి పనుల్లో అద్భుతంగా ఉంటాయి.
- **కోడ్ అసిస్టెంట్**. మీరు ఉపయోగించే భాషా మోడల్ ఆధారంగా, మీరు కోడ్ రాయడంలో సహాయపడే కోడ్ అసిస్టెంట్‌ను నిర్మించవచ్చు. ఉదాహరణకు, GitHub Copilot మరియు ChatGPT వంటి ఉత్పత్తులను ఉపయోగించి కోడ్ రాయడంలో సహాయం పొందవచ్చు.

## నేను ఎలా ప్రారంభించగలను?

సరే, మీరు సాధారణంగా LLMతో ఇంటిగ్రేట్ కావడానికి ఈ రెండు విధానాలు ఉంటాయి:

- API ఉపయోగించండి. ఇక్కడ మీరు మీ ప్రాంప్ట్‌తో వెబ్ రిక్వెస్టులను నిర్మించి, జనరేట్ చేసిన టెక్స్ట్‌ను తిరిగి పొందుతారు.
- లైబ్రరీ ఉపయోగించండి. లైబ్రరీలు API కాల్స్‌ను కప్పి వాటిని ఉపయోగించడాన్ని సులభతరం చేస్తాయి.

## లైబ్రరీలు/SDKలు

LLMలతో పని చేయడానికి కొన్ని ప్రసిద్ధ లైబ్రరీలు ఉన్నాయి:

- **openai**, ఈ లైబ్రరీ మీ మోడల్‌కు కనెక్ట్ కావడం మరియు ప్రాంప్ట్స్ పంపడం సులభం చేస్తుంది.

తర్వాత ఉన్నత స్థాయిలో పనిచేసే లైబ్రరీలు:

- **Langchain**. Langchain ప్రసిద్ధి పొందినది మరియు Python మద్దతు ఇస్తుంది.
- **Semantic Kernel**. Semantic Kernel మైక్రోసాఫ్ట్ యొక్క లైబ్రరీ, ఇది C#, Python, మరియు Java భాషలను మద్దతు ఇస్తుంది.

## openai ఉపయోగించి మొదటి యాప్

మనం ఎలా మొదటి యాప్‌ను నిర్మించాలో, ఏ లైబ్రరీలు అవసరమో, ఎంత అవసరమో చూద్దాం.

### openai ఇన్‌స్టాల్ చేయండి

  > [!NOTE] మీరు ఈ నోట్‌బుక్‌ను Codespaces లేదా Devcontainerలో నడుపుతున్నట్లయితే ఈ దశ అవసరం లేదు


OpenAI లేదా Azure OpenAIతో ఇంటరాక్ట్ చేయడానికి అనేక లైబ్రరీలు ఉన్నాయి. మీరు C#, Python, JavaScript, Java మరియు మరిన్ని ప్రోగ్రామింగ్ భాషలను కూడా ఉపయోగించవచ్చు.  
మేము `openai` Python లైబ్రరీని ఎంచుకున్నాము, కాబట్టి `pip` ఉపయోగించి దాన్ని ఇన్‌స్టాల్ చేస్తాము.

```bash
pip install openai
```

మీరు ఈ నోట్‌బుక్‌ను Codespaces లేదా Dev Containerలో నడుపుతున్నట్లయితే కాకపోతే, మీ మెషీన్‌లో [Python](https://www.python.org/) ను కూడా ఇన్‌స్టాల్ చేయాలి.

### రిసోర్స్ సృష్టించండి మరియు మీ API కీని కనుగొనండి

మీకు ఇప్పటికీ చేయకపోతే, ఈ క్రింది దశలను అనుసరించండి:

- OpenAIలో ఖాతా సృష్టించండి <https://platform.openai.com/signup>.
- ఇప్పుడు, మీ API కీలు పొందండి <https://platform.openai.com/api-keys>. 

>[!NOTE]
> మీ API కీని మీ కోడ్ నుండి వేరుగా ఉంచడం మంచిది. మీరు వాతావరణ చరాలు ఉపయోగించి ఇది చేయవచ్చు.
> - మీ .env ఫైల్‌లో `OPENAI_KEY` వాతావరణ చరాన్ని మీ API కీకి సెట్ చేయండి. మీరు ఈ కోర్సు యొక్క గత వ్యాయామాలు పూర్తి చేసినట్లయితే, మీరు ఇప్పటికే సెట్ అయిపోయారు.
> - API కీ ఒకసారి మాత్రమే యాక్సెస్ చేయగలదని గమనించడం ముఖ్యం. కాబట్టి, అది సరిగ్గా కాపీ అయిందో లేదో నిర్ధారించుకోవాలి. అది పనిచేయకపోతే, కీని తొలగించి కొత్తదాన్ని సృష్టించడం మంచిది.

### OpenAI కాన్ఫిగరేషన్ సెటప్ చేయండి

మీరు OpenAI ఉపయోగిస్తుంటే, ఇక్కడ మీరు ఎలా కాన్ఫిగర్ చేసుకోవాలో ఉంది:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

పై కోడ్‌లో మేము ఈ క్రింది వాటిని సెట్ చేస్తున్నాము:

- `api_key`, ఇది OpenAI డాష్‌బోర్డ్‌లో కనుగొన్న మీ API కీ.
- `deployment`, ఇది మీ GPT వెర్షన్.

> [!NOTE]
> `os.environ` వాతావరణ చరాలను చదవడానికి ఉపయోగించే ఫంక్షన్. మీరు దీన్ని `OPENAI_API_KEY` వంటి వాతావరణ చరాలను చదవడానికి ఉపయోగించవచ్చు.

## టెక్స్ట్ జనరేట్ చేయండి

టెక్స్ట్ జనరేట్ చేయడానికి `chat.completion` క్లాస్ ఉపయోగిస్తారు. ఉదాహరణ ఇక్కడ ఉంది:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

పై కోడ్‌లో, మేము ఒక completion ఆబ్జెక్ట్ సృష్టించి, ఉపయోగించాలనుకున్న మోడల్ మరియు ప్రాంప్ట్‌ను అందిస్తున్నాము. తరువాత మేము జనరేట్ చేసిన టెక్స్ట్‌ను ప్రింట్ చేస్తున్నాము.

### చాట్ కంప్లీషన్స్

ఇప్పటివరకు, మీరు `Completion` ఉపయోగించి టెక్స్ట్ ఎలా జనరేట్ చేస్తున్నామో చూశారు. కానీ `ChatCompletion` అనే మరో క్లాస్ ఉంది, ఇది చాట్‌బాట్స్‌కు మరింత అనుకూలంగా ఉంటుంది. దీన్ని ఉపయోగించే ఉదాహరణ ఇక్కడ ఉంది:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

### ఫిల్టర్ మరియు షాపింగ్ లిస్ట్ జోడించడం ద్వారా మెరుగుపరచండి

మనం ఇప్పుడు రెసిపీలు ఉత్పత్తి చేయగల ఒక పని చేసే యాప్‌ను కలిగి ఉన్నాము మరియు ఇది వినియోగదారుని ఇన్‌పుట్‌లపై ఆధారపడి ఉండటం వల్ల సౌలభ్యంగా ఉంటుంది, రెసిపీల సంఖ్య మాత్రమే కాకుండా ఉపయోగించిన పదార్థాలపై కూడా.

దాన్ని మరింత మెరుగుపరచడానికి, మనం క్రింది వాటిని జోడించాలనుకుంటున్నాము:

- **పదార్థాలను ఫిల్టర్ చేయండి**. మనకు ఇష్టంలేని లేదా అలెర్జీ ఉన్న పదార్థాలను ఫిల్టర్ చేయగలగాలి. ఈ మార్పును సాధించడానికి, మనం ఉన్న ప్రాంప్ట్‌ను సవరించి చివరలో ఫిల్టర్ షరతును ఇలా జోడించవచ్చు:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```
