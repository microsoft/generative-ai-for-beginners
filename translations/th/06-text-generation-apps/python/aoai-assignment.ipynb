# สร้างแอปสร้างข้อความ

คุณได้เห็นจนถึงตอนนี้ในหลักสูตรว่ามีแนวคิดหลักอย่างการใช้ prompt และแม้กระทั่งมีสาขาวิชาเฉพาะที่เรียกว่า "การออกแบบ prompt" เครื่องมือหลายอย่างที่คุณสามารถโต้ตอบได้ เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่น ๆ สนับสนุนให้คุณใช้ prompt เพื่อทำสิ่งต่าง ๆ

เพื่อให้คุณเพิ่มประสบการณ์แบบนี้ในแอป คุณจำเป็นต้องเข้าใจแนวคิดอย่าง prompt, completions และเลือกไลบรารีที่จะใช้ นั่นคือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## แนะนำ

ในบทนี้ คุณจะ:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปสร้างข้อความโดยใช้ openai
- เข้าใจวิธีการใช้แนวคิดอย่าง prompt, temperature และ tokens เพื่อสร้างแอปสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้ คุณจะสามารถ:

- อธิบายว่าแอปสร้างข้อความคืออะไร
- สร้างแอปสร้างข้อความโดยใช้ openai
- กำหนดค่าแอปของคุณให้ใช้ tokens มากกว่าหรือน้อยกว่า และปรับเปลี่ยนอุณหภูมิ เพื่อให้ได้ผลลัพธ์ที่หลากหลาย

## แอปสร้างข้อความคืออะไร?

ปกติเมื่อคุณสร้างแอป จะมีอินเทอร์เฟสบางอย่างเช่นตัวอย่างต่อไปนี้:

- แบบใช้คำสั่ง Console แอปประเภทนี้จะพิมพ์คำสั่งและดำเนินงานตามคำสั่ง ตัวอย่างเช่น `git` คือแอปแบบใช้คำสั่ง
- แบบมีอินเทอร์เฟสผู้ใช้ (UI) บางแอปมีอินเทอร์เฟสแบบกราฟิก GUI ที่คุณคลิกปุ่ม, กรอกข้อความ, เลือกตัวเลือกและอื่น ๆ

### แอป Console และ UI มีข้อจำกัด

เปรียบเทียบกับแอปแบบใช้คำสั่ง ที่คุณต้องพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีแค่คำสั่งที่แอปรองรับเท่านั้น
- **เฉพาะภาษา** บางแอปรองรับหลายภาษา แต่โดยทั่วไปแอปจะถูกสร้างมาสำหรับภาษาหนึ่ง แม้ว่าคุณจะสามารถเพิ่มการรองรับภาษาอื่นได้

### ข้อดีของแอปสร้างข้อความ

แล้วแอปสร้างข้อความแตกต่างอย่างไร?

ในแอปสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น คุณไม่ถูกจำกัดด้วยชุดคำสั่งหรือภาษาที่ใช้ป้อนข้อมูลเฉพาะ แต่สามารถใช้ภาษาธรรมชาติในการโต้ตอบกับแอป อีกข้อดีคือเพราะคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ผ่านการฝึกอบรมบนคอลเลกชันข้อมูลจำนวนมาก ต่างจากแอปแบบดั้งเดิมที่อาจจำกัดข้อมูลในฐานข้อมูล

### ฉันสามารถสร้างอะไรได้บ้างด้วยแอปสร้างข้อความ?

มีหลายสิ่งที่คุณสามารถสร้างได้ เช่น:

- **แชทบอท** แชทบอทที่ตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและผลิตภัณฑ์ จะเหมาะมาก
- **ผู้ช่วย** LLMs ดีมากในการทำสิ่งต่าง ๆ เช่น สรุปข้อความ, ดึงข้อมูลเชิงลึกจากข้อความ, ผลิตข้อความเช่นเรซูเม่ และอื่น ๆ
- **ผู้ช่วยโค้ด** ขึ้นอยู่กับโมเดลภาษา คุณสามารถสร้างผู้ช่วยเขียนโค้ดได้ เช่น คุณสามารถใช้ผลิตภัณฑ์อย่าง GitHub Copilot และ ChatGPT เพื่อช่วยเขียนโค้ด

### ตั้งค่าการกำหนดค่า Azure

ถ้าคุณใช้ Azure OpenAI นี่คือวิธีตั้งค่าการกำหนดค่า Responses API ให้บริการจาก Azure OpenAI **v1 endpoint** ดังนั้นเราชี้ให้ไคลเอนต์ `OpenAI` ไปที่ `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ข้างต้นเรากำลังตั้งค่าดังนี้:

- `api_key` คือตัวคีย์ API ของคุณที่หามาใน Azure Portal
- `base_url` คือ endpoint ของ Azure OpenAI พร้อมกับต่อท้าย `/openai/v1/` คุณสามารถหา endpoint นี้ใน Azure Portal ที่อยู่ใกล้กับคีย์ API ของคุณ การใช้ endpoint v1 หมายความว่าคุณไม่ต้องระบุ `api_version` อีกต่อไป
- `deployment` คือชื่อของการปรับใช้โมเดลที่คุณสร้างในพอร์ทัล Foundry

> [!NOTE]
> `os.environ` คือ mapping สำหรับอ่าน environment variables คุณสามารถใช้มันเพื่ออ่าน environment variables เช่น `AZURE_OPENAI_API_KEY` และ `AZURE_OPENAI_ENDPOINT`

## สร้างข้อความ

วิธีสร้างข้อความคือใช้เมธอด `responses.create` ตัวอย่างเช่น:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ในโค้ดข้างต้น เราสร้างอ็อบเจกต์ response และส่งโมเดลที่ต้องการใช้พร้อมกับ prompt จากนั้นพิมพ์ข้อความที่สร้างขึ้น

### การตอบในแบบแชท

Responses API เหมาะสำหรับทั้งการสร้างข้อความแบบรอบเดียวและแชทบอทแบบหลายรอบ - คุณเพียงแค่ส่งข้อความเพิ่มในลิสต์ `input` เพื่อสร้างการสนทนา:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

รายละเอียดเพิ่มเติมเกี่ยวกับฟังก์ชันนี้ในบทถัดไป

````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## ประเภทของ prompt ที่แตกต่างกัน สำหรับสิ่งที่แตกต่างกัน

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้ prompt แล้ว คุณยังมีโปรแกรมที่ทำงานได้ซึ่งคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่าง ๆ

Prompt สามารถใช้กับงานต่าง ๆ ได้มากมาย ตัวอย่างเช่น:

- **สร้างข้อความประเภทหนึ่ง** เช่น คุณสามารถสร้างบทกวี, คำถามสำหรับแบบทดสอบ ฯลฯ
- **ค้นหาข้อมูล** คุณสามารถใช้ prompt เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในการพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้ prompt เพื่อสร้างโค้ด เช่น พัฒนากฎ regular expression สำหรับตรวจสอบอีเมล หรือจะสร้างโปรแกรมทั้งหมด เช่น เว็บแอปก็ได้
