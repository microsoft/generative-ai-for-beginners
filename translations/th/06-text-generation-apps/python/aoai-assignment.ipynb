# สร้างแอปสำหรับสร้างข้อความ

คุณได้เห็นจนถึงตอนนี้จากหลักสูตรนี้ว่ามีแนวคิดหลักอย่างเช่น prompt และแม้กระทั่งสาขาวิชาเต็มที่เรียกว่า "prompt engineering" เครื่องมือมากมายที่คุณสามารถโต้ตอบด้วยได้ เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่นๆ ยังช่วยให้คุณใช้ prompt เพื่อทำบางสิ่งให้สำเร็จ

เพื่อให้คุณเพิ่มประสบการณ์แบบนี้ลงในแอป คุณจำเป็นต้องเข้าใจแนวคิดอย่าง prompt, completions และเลือกไลบรารีที่จะใช้งาน นั่นคือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## บทนำ

ในบทนี้ คุณจะ:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปสำหรับสร้างข้อความโดยใช้ openai
- เข้าใจวิธีใช้แนวคิดอย่าง prompt, temperature และ tokens เพื่อสร้างแอปสำหรับสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้ คุณจะสามารถ:

- อธิบายว่าแอปสำหรับสร้างข้อความคืออะไร
- สร้างแอปสำหรับสร้างข้อความโดยใช้ openai
- ปรับแต่งแอปของคุณเพื่อใช้ token มากขึ้นหรือน้อยลง และเปลี่ยน temperature เพื่อให้ได้ผลลัพธ์ที่หลากหลาย

## แอปสำหรับสร้างข้อความคืออะไร?

ปกติเมื่อคุณสร้างแอป มันจะมีอินเทอร์เฟซชนิดหนึ่งดังนี้:

- แบบคำสั่ง (Command-based) แอปคอนโซลมักเป็นแอปที่คุณพิมพ์คำสั่งแล้วมันจะทำงาน เช่น `git` คือแอปแบบคำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) แอปบางตัวมีส่วนติดต่อผู้ใช้แบบกราฟิก (GUI) ที่คุณคลิกปุ่ม ป้อนข้อความ เลือกตัวเลือก และอื่นๆ

### แอปคอนโซลและ UI มีข้อจำกัด

เปรียบเทียบกับแอปแบบคำสั่งที่คุณพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีเพียงคำสั่งที่แอปรองรับเท่านั้น
- **เฉพาะภาษาหนึ่งๆ** แอปบางตัวรองรับหลายภาษา แต่โดยค่าเริ่มต้นแอปนั้นถูกสร้างสำหรับภาษาหนึ่งแม้ว่าคุณจะสามารถเพิ่มการรองรับภาษาอื่นได้

### ประโยชน์ของแอปสำหรับสร้างข้อความ

แล้วแอปสำหรับสร้างข้อความต่างกันอย่างไร?

ในแอปสำหรับสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น ไม่จำกัดเฉพาะชุดคำสั่งหรือภาษาป้อนเฉพาะ คุณสามารถใช้ภาษาธรรมชาติเพื่อโต้ตอบกับแอป อีกประโยชน์หนึ่งคือ เนื่องจากคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ได้รับการฝึกฝนจากข้อมูลจำนวนมาก ต่างจากแอปแบบดั้งเดิมที่มักจำกัดอยู่ในฐานข้อมูล

### ฉันสามารถสร้างอะไรได้บ้างด้วยแอปสำหรับสร้างข้อความ?

มีหลายอย่างที่คุณสามารถสร้างได้ เช่น:

- **แชทบอต** แชทบอตที่ตอบคำถามเกี่ยวกับหัวข้อต่างๆ เช่น บริษัทและผลิตภัณฑ์ของคุณ อาจเป็นตัวเลือกที่ดี
- **ผู้ช่วย** LLM ดีเยี่ยมในการสรุปข้อความ รับข้อมูลเชิงลึกจากข้อความ สร้างข้อความอย่างเรซูเม่ และอื่นๆ
- **ผู้ช่วยเขียนโค้ด** ขึ้นอยู่กับโมเดลภาษา คุณสามารถสร้างผู้ช่วยเขียนโค้ดที่ช่วยเขียนโค้ดได้ เช่น ใช้ GitHub Copilot หรือ ChatGPT เพื่อช่วยเขียนโค้ด

- ใช้ API ที่นี่คุณกำลังสร้างคำขอเว็บด้วย prompt และได้รับข้อความที่สร้างกลับมา
- ใช้ไลบรารี ไลบรารีช่วยห่อหุ้มการเรียก API และทำให้ง่ายต่อการใช้งาน

### หา API key และ endpoint

ตอนนี้ คุณต้องแจ้งให้ไลบรารี `openai` ของคุณทราบว่าจะใช้ API key อะไร วิธีหาคือไปที่ส่วน "Keys and Endpoint" ของทรัพยากร Azure OpenAI ของคุณและคัดลอกค่าของ "Key 1"

ตอนนี้ที่คุณได้คัดลอกข้อมูลนี้แล้ว ให้เราบอกไลบรารีให้ใช้ค่าเหล่านี้

> [!NOTE]
> ควรแยก API key ออกจากโค้ดของคุณ คุณสามารถทำได้โดยใช้ตัวแปรแวดล้อม
> - ตั้งค่าตัวแปรแวดล้อม `AZURE_OPENAI_API_KEY` เป็น API key ของคุณในไฟล์ .env ถ้าคุณทำแบบฝึกหัดก่อนหน้านี้ของคอร์สนี้เสร็จแล้ว คุณก็พร้อมใช้งาน


### ตั้งค่าการกำหนดค่า Azure

ถ้าคุณใช้ Azure OpenAI นี่คือวิธีตั้งค่าการกำหนดค่า API Responses โดยให้บริการจาก Azure OpenAI ที่ **v1 endpoint** ดังนั้นเราจึงชี้ไคลเอนต์ `OpenAI` ไปที่ `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ข้างต้นเราได้ตั้งค่าดังนี้:

- `api_key` นี่คือ API key ของคุณที่หาได้จาก Azure Portal
- `base_url` นี่คือ endpoint ของ Azure OpenAI ของคุณที่มี `/openai/v1/` ต่อท้าย คุณสามารถหา endpoint นี้ได้ใน Azure Portal ข้าง API key การใช้ v1 endpoint หมายความว่าคุณไม่จำเป็นต้องส่ง `api_version`
- `deployment` นี่คือชื่อของโมเดล deployment ที่คุณสร้างใน Foundry portal

> [!NOTE]
> `os.environ` คือ mapping ที่อ่านค่าตัวแปรแวดล้อม คุณสามารถใช้เพื่ออ่านตัวแปรแวดล้อมเช่น `AZURE_OPENAI_API_KEY` และ `AZURE_OPENAI_ENDPOINT`

## สร้างข้อความ

วิธีสร้างข้อความคือใช้เมธอด `responses.create` ตัวอย่างเช่น:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ในโค้ดด้านบน เราสร้างอ็อบเจ็กต์ response และส่งโมเดลที่อยากใช้และ prompt ลงไป จากนั้นเราพิมพ์ข้อความที่สร้างขึ้น

### การตอบสนองแบบแชท

Responses API เหมาะสำหรับทั้งการสร้างข้อความรอบเดียวและแชทบอตหลายรอบ — คุณเพียงแค่ใส่ข้อความมากขึ้นในลิสต์ `input` เพื่อสร้างบทสนทนา:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

รายละเอียดเพิ่มเติมเกี่ยวกับฟังก์ชันนี้ในบทต่อไป

## แบบฝึกหัด - แอปสร้างข้อความแรกของคุณ

ตอนนี้ที่เราได้เรียนรู้วิธีตั้งค่าและกำหนดค่า Azure OpenAI แล้ว ถึงเวลาสร้างแอปสร้างข้อความแรกของคุณ เพื่อสร้างแอป ให้ทำตามขั้นตอนเหล่านี้:


1. สร้างสภาพแวดล้อมเสมือนและติดตั้ง openai:

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นถ้าคุณรันโน้ตบุ๊กนี้บน Codespaces หรือใน Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!หมายเหตุ]
> หากคุณใช้ Windows ให้พิมพ์ `venv\Scripts\activate` แทน `source venv/bin/activate`

> [!หมายเหตุ]
> ค้นหาคีย์ Azure OpenAI ของคุณโดยไปที่ https://portal.azure.com/ แล้วค้นหา `Open AI` และเลือก `Open AI resource` จากนั้นเลือก `Keys and Endpoint` แล้วคัดลอกค่าของ `Key 1`


1. สร้างไฟล์ *app.py* แล้วใส่โค้ดดังต่อไปนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

## ประเภทของพรอมต์ที่แตกต่างกัน สำหรับสิ่งที่แตกต่างกัน

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้พรอมต์แล้ว คุณยังมีโปรแกรมที่ทำงานอยู่ซึ่งคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่าง ๆ ได้

พรอมต์สามารถใช้ได้กับงานหลากหลายประเภท เช่น:

- **สร้างข้อความประเภทหนึ่ง** ตัวอย่างเช่น คุณสามารถสร้างกลอน คำถามสำหรับแบบทดสอบ ฯลฯ

- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น พัฒนากระบวนทัศน์ปกติที่ใช้ตรวจสอบอีเมล หรือจะให้สร้างโปรแกรมทั้งโปรแกรม เช่น เว็บแอป ก็ได้

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    ```python
    old_prompt_result = response.output_text

    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    ```

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be.
