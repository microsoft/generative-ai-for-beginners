# สร้างแอปสำหรับสร้างข้อความ

จนถึงตอนนี้คุณได้เห็นในหลักสูตรนี้ว่ามีแนวคิดหลักเช่น prompts และยังมีสาขาวิชาเฉพาะที่เรียกว่า "prompt engineering" อีกด้วย เครื่องมือหลายอย่างที่คุณสามารถโต้ตอบด้วยได้ เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่น ๆ รองรับการใช้ prompts เพื่อทำงานบางอย่างให้สำเร็จ

สำหรับคุณที่จะเพิ่มประสบการณ์ดังกล่าวลงในแอป คุณจำเป็นต้องเข้าใจแนวคิดเช่น prompts, completions และเลือกไลบรารีที่ใช้ทำงาน นั่นคือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## บทนำ

ในบทนี้ คุณจะได้:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปสำหรับสร้างข้อความโดยใช้ openai
- เข้าใจวิธีการใช้แนวคิด เช่น prompt, temperature, และ tokens เพื่อสร้างแอปสำหรับสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้ คุณจะสามารถ:

- อธิบายว่าแอปสำหรับสร้างข้อความคืออะไร
- สร้างแอปสำหรับสร้างข้อความโดยใช้ openai
- กำหนดค่าแอปของคุณให้ใช้ tokens มากหรือน้อย และเปลี่ยน temperature เพื่อให้ได้ผลลัพธ์ที่หลากหลาย

## แอปสำหรับสร้างข้อความคืออะไร?

โดยปกติเมื่อคุณสร้างแอป จะมีอินเทอร์เฟซบางชนิด เช่น ต่อไปนี้:

- แบบใช้คำสั่ง (Command-based) แอปคอนโซลเป็นแอปทั่วไปที่คุณพิมพ์คำสั่งและมันจะทำงานตามคำสั่งนั้น ๆ เช่น `git` เป็นแอปแบบใช้คำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) บางแอปมีอินเทอร์เฟซแบบกราฟิก (GUI) ที่คุณคลิกปุ่ม ป้อนข้อความ เลือกตัวเลือก และอื่น ๆ

### แอปคอนโซลและ UI ถูกจำกัด

เปรียบเทียบกับแอปแบบใช้คำสั่งที่คุณพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีแค่คำสั่งที่แอปรองรับเท่านั้น
- **จำเพาะภาษาที่ใช้** บางแอปรองรับหลายภาษา แต่โดยปริยายแอปถูกสร้างมาเพื่อภาษาเฉพาะ แม้ว่าคุณจะสามารถเพิ่มการรองรับภาษาอื่น ๆ ได้ 

### ประโยชน์ของแอปสำหรับสร้างข้อความ

แล้วแอปสำหรับสร้างข้อความแตกต่างอย่างไร?

ในแอปสำหรับสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น คุณไม่ถูกจำกัดด้วยชุดคำสั่งหรืิอภาษาป้อนข้อมูลที่เฉพาะเจาะจง แต่คุณสามารถใช้ภาษาธรรมชาติเพื่อโต้ตอบกับแอป อีกประโยชน์คือเนื่องจากคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ถูกฝึกฝนจากข้อมูลจำนวนมาก ในขณะที่แอปแบบดั้งเดิมอาจถูกจำกัดด้วยข้อมูลในฐานข้อมูล

### ฉันสามารถสร้างอะไรได้บ้างกับแอปสำหรับสร้างข้อความ?

มีหลายอย่างที่คุณสามารถสร้าง เช่น:

- **แชทบอท** แชทบอทยังตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและผลิตภัณฑ์ของบริษัทได้เป็นอย่างดี
- **ผู้ช่วย** LLMs ดีมากในการสรุปข้อความ ดึงข้อมูลเชิงลึกจากข้อความ ผลิตข้อความเช่นเรซูเม่ และอื่น ๆ
- **ผู้ช่วยเขียนโค้ด** ขึ้นอยู่กับโมเดลภาษาที่คุณใช้ คุณสามารถสร้างผู้ช่วยเขียนโค้ดที่ช่วยเขียนโค้ดได้ ตัวอย่างเช่น คุณสามารถใช้ผลิตภัณฑ์อย่าง GitHub Copilot และ ChatGPT เพื่อช่วยเขียนโค้ด

## ฉันจะเริ่มต้นอย่างไร?

โดยทั่วไปคุณจำเป็นต้องหาวิธีการเชื่อมต่อกับ LLM ซึ่งโดยปกติจะมีสองวิธีดังนี้:

- ใช้ API คุณสร้างคำขอเว็บพร้อมกับ prompt ของคุณและรับข้อความที่สร้างขึ้นกลับมา
- ใช้ไลบรารี ไลบรารีช่วยห่อหุ้มการเรียก API และทำให้ง่ายต่อการใช้งาน

## ไลบรารี / SDKs

มีไลบรารีที่เป็นที่รู้จักจำนวนหนึ่งสำหรับการทำงานกับ LLMs เช่น:

- **openai** ไลบรารีนี้ทำให้ง่ายต่อการเชื่อมต่อกับโมเดลของคุณและส่ง prompts

จากนั้นก็มีไลบรารีที่ทำงานในระดับสูงขึ้น เช่น:

- **Langchain** Langchain เป็นที่รู้จักดีและรองรับ Python
- **Semantic Kernel** Semantic Kernel เป็นไลบรารีโดย Microsoft ที่รองรับภาษา C#, Python และ Java

## สร้างข้อความด้วย ChatCompletions

วิธีสร้างข้อความคือใช้คลาส `ChatCompletionsClient`
ใน `samples/python/azure_ai_inference/basic.py` ในส่วน code ตอบกลับ อัปเดตโค้ดส่วน user role โดยเปลี่ยนพารามิเตอร์ content เป็นดังนี้:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # พารามิเตอร์ที่เลือกใช้ได้
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

รันไฟล์ที่อัปเดตเพื่อดูผลลัพธ์


## ประเภทของ Prompt ที่แตกต่างกัน สำหรับสิ่งที่แตกต่างกัน

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้ prompt แล้ว คุณยังมีโปรแกรมที่ทำงานอยู่และคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่าง ๆ ได้

Prompt สามารถใช้สำหรับงานหลากหลายประเภท เช่น:

- **สร้างข้อความประเภทหนึ่ง** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ ฯลฯ
- **ค้นหาข้อมูล** คุณสามารถใช้ prompt เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในการพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้ prompt เพื่อสร้างโค้ด เช่น การพัฒนากฎ regular expression สำหรับตรวจสอบอีเมล หรือไม่ก็สร้างโปรแกรมทั้งหมด เช่น เว็บแอป

## แบบฝึกหัด - สร้างเครื่องสร้างสูตรอาหาร

ตอนนี้ที่เราได้เล่นสถานการณ์แล้ว เรามาเขียนโค้ดให้ตรงกับสถานการณ์ที่แสดงกัน ต่อไปนี้ทำตามขั้นตอนดังนี้:

1. ใช้ไฟล์ที่มีอยู่แล้วเป็นจุดเริ่มต้น
1. สร้างตัวแปร `prompt` และแก้ไขโค้ดตัวอย่างดังนี้:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

ดีมาก เรามาดูวิธีที่เราจะปรับปรุงสิ่งต่างๆ กัน เพื่อปรับปรุงสิ่งต่างๆ เราต้องการให้แน่ใจว่าโค้ดมีความยืดหยุ่น ดังนั้นวัตถุดิบและจำนวนสูตรอาหารจึงสามารถปรับปรุงและเปลี่ยนแปลงได้


1. มาลองเปลี่ยนโค้ดในลักษณะดังต่อไปนี้:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
