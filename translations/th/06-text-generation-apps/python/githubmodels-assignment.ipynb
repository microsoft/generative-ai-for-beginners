# สร้างแอปพลิเคชันสร้างข้อความ

ตลอดหลักสูตรนี้ คุณจะได้เห็นแนวคิดหลักอย่างเช่น prompt และแม้แต่ศาสตร์ที่เรียกว่า "prompt engineering" เครื่องมือหลายอย่างที่คุณใช้งานได้ เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่น ๆ ล้วนรองรับการใช้ prompt เพื่อให้คุณทำสิ่งต่าง ๆ ได้

ถ้าคุณอยากเพิ่มประสบการณ์แบบนี้ลงในแอปของคุณเอง คุณต้องเข้าใจแนวคิดอย่าง prompt, completion และเลือกไลบรารีที่เหมาะสม ซึ่งทั้งหมดนี้คุณจะได้เรียนรู้ในบทนี้

## บทนำ

ในบทนี้ คุณจะได้:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปพลิเคชันสร้างข้อความโดยใช้ openai
- เข้าใจวิธีใช้แนวคิดอย่าง prompt, temperature และ token เพื่อสร้างแอปพลิเคชันสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้ คุณจะสามารถ:

- อธิบายได้ว่าแอปพลิเคชันสร้างข้อความคืออะไร
- สร้างแอปพลิเคชันสร้างข้อความโดยใช้ openai
- ปรับแต่งแอปของคุณให้ใช้ token มากหรือน้อย และเปลี่ยนค่า temperature เพื่อให้ได้ผลลัพธ์ที่หลากหลาย

## แอปพลิเคชันสร้างข้อความคืออะไร?

ปกติแล้วเวลาเราสร้างแอป มักจะมีอินเทอร์เฟซแบบใดแบบหนึ่ง เช่น:

- แบบใช้คำสั่ง (Command-based) แอปคอนโซลเป็นตัวอย่างที่ดี ที่คุณพิมพ์คำสั่งแล้วแอปจะทำงาน เช่น `git` ก็เป็นแอปแบบใช้คำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) บางแอปมีอินเทอร์เฟซแบบกราฟิก (GUI) ที่คุณคลิกปุ่ม, พิมพ์ข้อความ, เลือกตัวเลือก ฯลฯ

### แอปคอนโซลและ UI มีข้อจำกัด

ลองเปรียบเทียบกับแอปแบบใช้คำสั่งที่คุณต้องพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ ต้องเป็นคำสั่งที่แอปรองรับเท่านั้น
- **จำกัดภาษาที่ใช้** บางแอปรองรับหลายภาษา แต่โดยปกติแล้วแอปจะถูกสร้างมาสำหรับภาษาใดภาษาหนึ่ง แม้จะเพิ่มการรองรับภาษาอื่นได้ก็ตาม

### ข้อดีของแอปพลิเคชันสร้างข้อความ

แล้วแอปพลิเคชันสร้างข้อความต่างจากนี้อย่างไร?

ในแอปพลิเคชันสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น ไม่จำกัดแค่ชุดคำสั่งหรือภาษาที่ป้อนเข้าไป คุณสามารถใช้ภาษาธรรมชาติในการโต้ตอบกับแอปได้ อีกข้อดีคือคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ผ่านการฝึกฝนจากข้อมูลจำนวนมหาศาล ในขณะที่แอปแบบเดิมอาจจำกัดแค่ข้อมูลในฐานข้อมูลเท่านั้น

### แอปพลิเคชันสร้างข้อความนำไปสร้างอะไรได้บ้าง?

มีหลายอย่างที่คุณสามารถสร้างได้ เช่น:

- **แชทบอท** แชทบอทที่ตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและผลิตภัณฑ์ของบริษัทก็เหมาะมาก
- **ผู้ช่วย** LLM เก่งในเรื่องการสรุปข้อความ, วิเคราะห์ข้อมูลจากข้อความ, สร้างข้อความเช่นเรซูเม่ ฯลฯ
- **ผู้ช่วยเขียนโค้ด** ขึ้นอยู่กับโมเดลภาษา คุณสามารถสร้างผู้ช่วยเขียนโค้ดที่ช่วยคุณเขียนโปรแกรมได้ เช่น ใช้ GitHub Copilot หรือ ChatGPT เพื่อช่วยเขียนโค้ด

## จะเริ่มต้นอย่างไรดี?

คุณต้องหาวิธีเชื่อมต่อกับ LLM ซึ่งโดยทั่วไปมี 2 วิธีหลัก ๆ คือ:

- ใช้ API คุณจะสร้าง web request พร้อม prompt ของคุณ แล้วรับข้อความที่สร้างกลับมา
- ใช้ไลบรารี ไลบรารีจะช่วยจัดการการเรียก API ให้ใช้งานง่ายขึ้น

## ไลบรารี/SDK

มีไลบรารีที่เป็นที่รู้จักสำหรับใช้งานกับ LLM เช่น:

- **openai** ไลบรารีนี้ช่วยให้เชื่อมต่อกับโมเดลของคุณและส่ง prompt ได้ง่าย

นอกจากนี้ยังมีไลบรารีที่ทำงานในระดับสูงขึ้น เช่น:

- **Langchain** เป็นไลบรารีที่ได้รับความนิยมและรองรับ Python
- **Semantic Kernel** เป็นไลบรารีของ Microsoft ที่รองรับภาษา C#, Python และ Java

## สร้างข้อความด้วย ChatCompletions

วิธีสร้างข้อความคือใช้คลาส `ChatCompletionsClient`  
ใน `samples/python/azure_ai_inference/basic.py` ในส่วน response ของโค้ด ให้แก้ไขโค้ดในส่วน user role โดยเปลี่ยนค่า content ตามด้านล่างนี้:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

รันไฟล์ที่อัปเดตแล้วเพื่อดูผลลัพธ์


## ประเภทของพรอมต์ที่แตกต่างกัน สำหรับงานที่แตกต่างกัน

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความด้วยพรอมต์แล้ว คุณยังมีโปรแกรมที่สามารถปรับแต่งและเปลี่ยนแปลงเพื่อสร้างข้อความในรูปแบบต่าง ๆ ได้อีกด้วย

พรอมต์สามารถนำไปใช้กับงานหลากหลายประเภท ตัวอย่างเช่น:

- **สร้างข้อความในรูปแบบต่าง ๆ** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ ฯลฯ
- **ค้นหาข้อมูล** คุณสามารถใช้พรอมต์เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในงานพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น การพัฒนา regular expression สำหรับตรวจสอบอีเมล หรือแม้แต่สร้างโปรแกรมทั้งโปรแกรม เช่น เว็บแอป

## แบบฝึกหัด - สร้างตัวสร้างสูตรอาหาร

ตอนนี้เราได้ลองเล่นสถานการณ์ตัวอย่างไปแล้ว มาลองเขียนโค้ดให้ตรงกับสถานการณ์ที่แสดงไว้กันเถอะ ทำตามขั้นตอนเหล่านี้:

1. ใช้ไฟล์ที่มีอยู่เป็นจุดเริ่มต้น
1. สร้างตัวแปร `prompt` และปรับโค้ดตัวอย่างตามด้านล่างนี้:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

เยี่ยมเลย มาดูกันว่าเราจะปรับปรุงอะไรได้บ้าง เพื่อให้ดีขึ้น เราต้องการให้โค้ดยืดหยุ่นมากขึ้น เพื่อที่วัตถุดิบและจำนวนสูตรอาหารจะสามารถปรับเปลี่ยนและเพิ่มเติมได้


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

**ข้อจำกัดความรับผิดชอบ**:  
เอกสารฉบับนี้ได้รับการแปลโดยใช้บริการแปลภาษา AI [Co-op Translator](https://github.com/Azure/co-op-translator) แม้ว่าเราจะพยายามอย่างเต็มที่เพื่อความถูกต้อง แต่โปรดทราบว่าการแปลโดยระบบอัตโนมัติอาจมีข้อผิดพลาดหรือความไม่ถูกต้อง เอกสารต้นฉบับในภาษาต้นทางควรถือเป็นแหล่งข้อมูลที่เชื่อถือได้ สำหรับข้อมูลที่มีความสำคัญ แนะนำให้ใช้บริการแปลโดยนักแปลมืออาชีพ ทางเราจะไม่รับผิดชอบต่อความเข้าใจผิดหรือการตีความที่คลาดเคลื่อนซึ่งเกิดจากการใช้การแปลนี้
