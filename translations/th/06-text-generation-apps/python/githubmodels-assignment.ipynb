# สร้างแอปพลิเคชันสร้างข้อความ

คุณได้เห็นจนถึงตอนนี้ผ่านหลักสูตรนี้ว่า มีแนวคิดหลักอย่างเช่น prompt และแม้แต่สาขาทั้งหมดที่เรียกว่า "prompt engineering" เครื่องมือหลายตัวที่คุณสามารถโต้ตอบด้วย เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่น ๆ รองรับการที่คุณใช้ prompt เพื่อทำบางสิ่งให้สำเร็จ

เพื่อให้คุณเพิ่มประสบการณ์ดังกล่าวลงในแอป คุณจำเป็นต้องเข้าใจแนวคิดเช่น prompt, completions และเลือกไลบรารีที่ใช้ทำงาน นี่คือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## บทนำ

ในบทนี้ คุณจะได้:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปพลิเคชันสร้างข้อความโดยใช้ openai
- เข้าใจวิธีใช้แนวคิดอย่าง prompt, temperature, และ tokens ในการสร้างแอปพลิเคชันสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้แล้ว คุณจะสามารถ:

- อธิบายว่าแอปพลิเคชันสร้างข้อความคืออะไร
- สร้างแอปพลิเคชันสร้างข้อความโดยใช้ openai
- กำหนดค่าแอปของคุณให้ใช้ token มากขึ้นหรือน้อยลง และเปลี่ยนอุณหภูมิ เพื่อผลลัพธ์ที่หลากหลาย

## แอปพลิเคชันสร้างข้อความคืออะไร?

ปกติเมื่อคุณสร้างแอป จะมีอินเทอร์เฟซบางประเภทเช่นนี้:

- แบบคำสั่ง (Command-based) แอปคอนโซลเป็นแอปแบบที่คุณพิมพ์คำสั่งและมันทำงาน ตัวอย่างเช่น `git` คือแอปแบบคำสั่ง
- อินเทอร์เฟซผู้ใช้ (User interface, UI) บางแอปมีอินเทอร์เฟซกราฟิก (GUI) ที่คุณคลิกปุ่ม, กรอกข้อความ, เลือกตัวเลือก และอื่น ๆ

### แอปคอนโซลและ UI มีข้อจำกัด

เปรียบเทียบกับแอปแบบคำสั่งที่คุณพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีเพียงคำสั่งที่แอปรองรับเท่านั้น
- **จำเพาะตามภาษา** บางแอปรองรับหลายภาษา แต่โดยปกติแอปจะสร้างสำหรับภาษาหนึ่งโดยเฉพาะ ถึงแม้ว่าคุณจะสามารถเพิ่มการรองรับภาษาอื่นได้

### ข้อดีของแอปสร้างข้อความ

แล้วแอปสร้างข้อความต่างกันอย่างไร?

ในแอปสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น ไม่จำกัดแค่ชุดคำสั่งหรือภาษาป้อนเข้าที่เฉพาะเจาะจง แต่คุณสามารถใช้ภาษาธรรมชาติเพื่อโต้ตอบกับแอป อีกข้อดีคือ เพราะคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ผ่านการฝึกฝนกับข้อมูลจำนวนมาก ต่างจากแอปแบบดั้งเดิมที่อาจจำกัดว่ามีข้อมูลอะไรในฐานข้อมูลเท่านั้น

### ฉันสามารถสร้างอะไรได้บ้างกับแอปสร้างข้อความ?

คุณสามารถสร้างได้หลายอย่าง ตัวอย่างเช่น:

- **แชทบอท** แชทบอทตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทคุณและผลิตภัณฑ์ เช่นนี้เหมาะมาก
- **ผู้ช่วย** LLMs เก่งในการสรุปข้อความ, ได้ข้อมูลเชิงลึกจากข้อความ, สร้างข้อความเช่นประวัติย่อ และอื่น ๆ
- **ผู้ช่วยโค้ด** ขึ้นอยู่กับโมเดลภาษาที่คุณใช้ คุณสามารถสร้างผู้ช่วยเขียนโค้ด เช่น ใช้ผลิตภัณฑ์อย่าง GitHub Copilot และ ChatGPT เพื่อช่วยเขียนโค้ดได้

## ฉันจะเริ่มต้นอย่างไร?

คุณต้องหาวิธีเชื่อมต่อกับ LLM ซึ่งโดยทั่วไปมีสองวิธีนี้:

- ใช้ API ที่นี่คุณจะสร้างคำขอเว็บด้วย prompt ของคุณและรับข้อความที่สร้างกลับมา
- ใช้ไลบรารี ไลบรารีช่วยห่อหุ้มการเรียก API ทำให้ง่ายต่อการใช้

## ไลบรารี/SDK

มีไลบรารีที่รู้จักดีสำหรับทำงานกับ LLM หลายตัว เช่น:

- **openai** ไลบรารีนี้ทำให้ง่ายต่อการเชื่อมต่อกับโมเดลของคุณและส่ง prompt เข้าไป

และยังมีไลบรารีที่ทำงานในระดับสูงกว่า เช่น:

- **Langchain** Langchain มีชื่อเสียงและรองรับ Python
- **Semantic Kernel** Semantic Kernel เป็นไลบรารีของ Microsoft รองรับภาษา C#, Python, และ Java

## สร้างข้อความด้วย ChatCompletions

วิธีสร้างข้อความคือใช้คลาส `ChatCompletionsClient`
ใน `samples/python/azure_ai_inference/basic.py` ในส่วนตอบสนองของโค้ด ให้ปรับเปลี่ยนโค้ดผู้ใช้ role โดยเปลี่ยนพารามิเตอร์ content ดังนี้:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # พารามิเตอร์ที่ไม่จำเป็นต้องใส่
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

รันไฟล์ที่อัปเดตเพื่อดูผลลัพธ์


## ประเภทต่างๆ ของพรอมต์ สำหรับสิ่งที่แตกต่างกัน

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้พรอมต์แล้ว คุณยังมีโปรแกรมที่กำลังทำงานซึ่งคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่างๆ ได้

พรอมต์สามารถใช้ได้กับงานต่างๆ มากมาย เช่น:

- **สร้างข้อความประเภทหนึ่ง** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ เป็นต้น
- **ค้นหาข้อมูล** คุณสามารถใช้พรอมต์เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในการพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น การพัฒนากลุ่มนิพจน์ปกติที่ใช้ตรวจสอบอีเมล หรือจะสร้างโปรแกรมทั้งหมด เช่น เว็บแอปก็ได้

## แบบฝึกหัด - สร้างตัวสร้างสูตรอาหาร

ตอนนี้ที่เราได้เล่นสถานการณ์ไปแล้ว มาเขียนโค้ดให้ตรงกับสถานการณ์ที่แสดงกันเถอะ เพื่อทำเช่นนั้น ให้ทำตามขั้นตอนเหล่านี้:

1. ใช้ไฟล์ที่มีอยู่แล้วเป็นจุดเริ่มต้น
1. สร้างตัวแปร `prompt` และเปลี่ยนตัวอย่างโค้ดตามด้านล่างนี้:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

ดีมาก งั้นมาดูกันว่าวิธีที่เราสามารถปรับปรุงสิ่งต่าง ๆ ได้อย่างไร เพื่อปรับปรุงสิ่งต่าง ๆ เราต้องการให้โค้ดยืดหยุ่น ดังนั้นวัตถุดิบและจำนวนสูตรอาหารสามารถปรับปรุงและเปลี่ยนแปลงได้


1. มาเปลี่ยนโค้ดดังนี้:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
