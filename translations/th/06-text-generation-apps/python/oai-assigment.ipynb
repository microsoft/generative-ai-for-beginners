# สร้างแอปสร้างข้อความ

คุณได้เห็นมาแล้วในหลักสูตรนี้ว่ามีแนวคิดหลักอย่าง prompts และแม้แต่สาขาวิชาเต็มรูปแบบที่เรียกว่า "prompt engineering" เครื่องมือหลายอย่างที่คุณสามารถโต้ตอบได้ เช่น ChatGPT, Office 365, Microsoft Power Platform และอื่น ๆ สนับสนุนให้คุณใช้ prompts เพื่อทำบางอย่างให้สำเร็จ

เพื่อที่คุณจะเพิ่มประสบการณ์เช่นนี้ในแอป คุณต้องเข้าใจแนวคิดอย่าง prompts, completions และเลือกไลบรารีที่จะใช้งาน นั่นคือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## บทนำ

ในบทนี้คุณจะได้:

- เรียนรู้เกี่ยวกับไลบรารี openai และแนวคิดหลักของมัน
- สร้างแอปสร้างข้อความโดยใช้ openai
- เข้าใจวิธีใช้แนวคิดอย่าง prompt, temperature และ tokens เพื่อสร้างแอปสร้างข้อความ

## เป้าหมายการเรียนรู้

เมื่อสิ้นสุดบทเรียนนี้ คุณจะสามารถ:

- อธิบายว่าแอปสร้างข้อความคืออะไร
- สร้างแอปสร้างข้อความโดยใช้ openai
- กำหนดค่าแอปของคุณให้ใช้ tokens มากหรือน้อยลง และเปลี่ยนอุณหภูมิ เพื่อให้ได้ผลลัพธ์ที่หลากหลาย

## แอปสร้างข้อความคืออะไร?

โดยปกติเมื่อคุณสร้างแอป จะมีอินเทอร์เฟซบางประเภทดังนี้:

- แบบคำสั่ง (Command-based) แอปคอนโซลเป็นแอปทั่วไปที่คุณพิมพ์คำสั่งและมันทำงาน เช่น `git` เป็นแอปแบบคำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) บางแอปมีอินเทอร์เฟซกราฟิกผู้ใช้ (GUI) ที่คุณคลิกปุ่ม ป้อนข้อความ เลือกตัวเลือก และอื่น ๆ

### แอปคอนโซลและ UI มีข้อจำกัด

เปรียบเทียบกับแอปแบบคำสั่งที่คุณพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีแค่คำสั่งที่แอปรองรับเท่านั้น
- **เฉพาะภาษา** บางแอปรองรับหลายภาษา แต่โดยค่าเริ่มต้นแอปถูกสร้างสำหรับภาษาหนึ่งภาษา แม้คุณจะสามารถเพิ่มการรองรับภาษาอื่นได้

### ข้อดีของแอปสร้างข้อความ

แล้วแอปสร้างข้อความแตกต่างอย่างไร?

ในแอปสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น คุณไม่ได้จำกัดอยู่แค่ชุดคำสั่งหรือภาษาอินพุตเฉพาะ แต่คุณสามารถใช้ภาษาธรรมชาติโต้ตอบกับแอป อีกประโยชน์คือเพราะคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ได้รับการฝึกฝนจากข้อมูลมหาศาล ต่างจากแอปแบบดั้งเดิมที่อาจจำกัดเฉพาะข้อมูลในฐานข้อมูล

### ฉันสามารถสร้างอะไรได้บ้างกับแอปสร้างข้อความ?

มีหลายสิ่งที่คุณสามารถสร้างได้ เช่น:

- **แชทบอท** แชทบอทที่ตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและผลิตภัณฑ์ของบริษัทน่าจะเหมาะสม
- **ผู้ช่วย** LLMs เหมาะกับงานอย่างการสรุปข้อความ การดึงข้อมูลเชิงลึกจากข้อความ การสร้างข้อความอย่างเรซูเม่ และอื่น ๆ
- **ผู้ช่วยเขียนโค้ด** ขึ้นอยู่กับโมเดลภาษาที่คุณใช้ คุณสามารถสร้างผู้ช่วยเขียนโค้ดช่วยคุณเขียนโค้ด เช่น คุณสามารถใช้ผลิตภัณฑ์อย่าง GitHub Copilot หรือ ChatGPT เพื่อช่วยเขียนโค้ด

## ฉันจะเริ่มต้นอย่างไร?

คุณต้องหาวิธีเชื่อมต่อกับ LLM ซึ่งโดยทั่วไปมีสองวิธีดังนี้:

- ใช้ API ที่นี่คุณจะสร้างคำขอเว็บด้วย prompt ของคุณและรับข้อความที่ถูกสร้างกลับมา
- ใช้ไลบรารี ไลบรารีช่วยห่อหุ้มการเรียก API และทำให้ง่ายต่อการใช้งาน

## แอปแรกโดยใช้ openai

มาดูกันว่าคุณจะสร้างแอปแรกของคุณอย่างไร ไลบรารีที่ต้องใช้คืออะไร ต้องเตรียมตัวมากน้อยแค่ไหน และอื่น ๆ

### ติดตั้ง openai

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นหากคุณรันโน้ตบุ๊กนี้บน Codespaces หรือใน Devcontainer


มีไลบรารีหลายตัวสำหรับโต้ตอบกับ OpenAI หรือ Azure OpenAI และสามารถใช้ภาษาการเขียนโปรแกรมหลายภาษา เช่น C#, Python, JavaScript, Java และอื่น ๆ  
เราเลือกใช้ไลบรารี `openai` ของ Python ดังนั้นเราจะใช้ `pip` ในการติดตั้ง

```bash
pip install openai
```

หากคุณไม่ได้รันโน้ตบุ๊กนี้ใน Codespaces หรือ Dev Container คุณต้องติดตั้ง [Python](https://www.python.org/) ในเครื่องของคุณด้วย

### สร้างทรัพยากรและหากุญแจ API ของคุณ

ในกรณีที่คุณยังไม่ได้ทำ คุณต้องทำตามขั้นตอนดังนี้:

- สร้างบัญชีที่ OpenAI <https://platform.openai.com/signup>
- ตอนนี้ รับกุญแจ API ของคุณที่ <https://platform.openai.com/api-keys> 

>[!NOTE]
> ควรแยกกุญแจ API ของคุณออกจากโค้ด คุณสามารถทำได้โดยใช้ตัวแปรแวดล้อม
> - กำหนดตัวแปรแวดล้อมชื่อ `OPENAI_KEY` เป็นกุญแจ API ของคุณในไฟล์ .env หากคุณทำแบบฝึกหัดก่อนหน้านี้แล้ว คุณก็ตั้งค่านี้เรียบร้อย
> - สิ่งสำคัญคือต้องทราบว่ากุญแจ API จะเข้าถึงได้เพียงครั้งเดียว ดังนั้นจำเป็นต้องตรวจสอบให้แน่ใจว่าคัดลอกถูกต้อง หากไม่ทำงานตามที่ตั้งใจ แนะนำให้ลบกุญแจและสร้างกุญแจใหม่


### ตั้งค่าการกำหนดค่า OpenAI

หากคุณใช้ OpenAI นี่คือวิธีตั้งค่าการกำหนดค่า:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

ข้างบนนี้เรากำหนดค่า:

- `api_key` นี่คือกุญแจ API ของคุณที่พบในแดชบอร์ด OpenAI
- `deployment` นี่คือรุ่น GPT ของคุณ

> [!NOTE]
> `os.environ` เป็นฟังก์ชันที่อ่านตัวแปรแวดล้อม คุณสามารถใช้มันในการอ่านตัวแปรแวดล้อมอย่าง `OPENAI_API_KEY`

## สร้างข้อความ

วิธีการสร้างข้อความคือใช้เมธอด `responses.create` ดังตัวอย่างนี้:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ในโค้ดข้างต้น เราสร้างวัตถุตอบกลับและส่งโมเดลที่ต้องการใช้กับ prompt เข้าไป จากนั้นพิมพ์ข้อความที่ถูกสร้าง

### การตอบสนองแบบแชท

จนถึงตอนนี้ คุณเห็นแล้วว่าเราสามารถสร้างข้อความจาก prompt เดียวได้อย่างไร API Responses ยังเหมาะสำหรับแชทบอทด้วย - คุณจะให้รายการข้อความเป็น `input` เพื่อสร้างบทสนทนา นี่คือตัวอย่างการใช้:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

จะมีรายละเอียดเพิ่มเติมสำหรับฟังก์ชันนี้ในบทต่อไป

## แบบฝึกหัด - แอปสร้างข้อความแอปแรกของคุณ

ตอนนี้ที่เราเรียนรู้วิธีตั้งค่าและกำหนดค่า OpenAI แล้ว ถึงเวลาสร้างแอปสร้างข้อความแอปแรกของคุณ ทำตามขั้นตอนต่อไปนี้เพื่อสร้างแอปของคุณ:


1. สร้างสภาพแวดล้อมเสมือนและติดตั้ง openai:

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นหากคุณรันโน้ตบุ๊กนี้บน Codespaces หรือภายใน Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> หากคุณใช้ Windows ให้พิมพ์ `venv\Scripts\activate` แทน `source venv/bin/activate`

> [!NOTE]
> ค้นหาคีย์ OpenAI ของคุณโดยไปที่ https://platform.openai.com/settings/organization/api-keys และค้นหาคำว่า `API keys` คุณสามารถสร้างคีย์ใหม่ที่นั่นและคัดลอกค่าทันที


1. สร้างไฟล์ *app.py* และใส่โค้ดดังต่อไปนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````



## ประเภทต่าง ๆ ของพรอมต์ สำหรับสิ่งต่าง ๆ

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้พรอมต์แล้ว คุณยังมีโปรแกรมที่ใช้งานได้ซึ่งคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่าง ๆ

พรอมต์สามารถใช้สำหรับงานหลายประเภท เช่น:

- **สร้างข้อความประเภทหนึ่ง** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ ฯลฯ
- **ค้นหาข้อมูล** คุณสามารถใช้พรอมต์เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในการพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น พัฒนากลุ่มคำสั่งปกติที่ใช้ตรวจสอบอีเมล หรือจะสร้างโปรแกรมทั้งหมด เช่น เว็บแอปก็ได้

## แบบฝึกหัด - สร้างโปรแกรมสร้างสูตรอาหาร

ตอนนี้ที่เราได้เล่นสถานการณ์ไปแล้ว มาเขียนโค้ดให้ตรงกับสถานการณ์ที่แสดงกันเถอะ ในการทำเช่นนั้นให้ทำตามขั้นตอนดังนี้:

1. ใช้ไฟล์ *app.py* ที่มีอยู่เดิมเป็นจุดเริ่มต้น
1. หาตัวแปร `prompt` และเปลี่ยนโค้ดของมันเป็นดังต่อไปนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

ดีมาก ลองมาดูกันว่าทำอย่างไรเราจะปรับปรุงให้ดีขึ้น เพื่อปรับปรุงสิ่งต่าง ๆ เราต้องการให้โค้ดมีความยืดหยุ่น ดังนั้นส่วนผสมและจำนวนสูตรอาหารจึงสามารถปรับปรุงและเปลี่ยนแปลงได้

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### ปรับปรุงโดยการเพิ่มตัวกรองและรายการซื้อของ

ตอนนี้เรามีแอปที่ทำงานได้ซึ่งสามารถผลิตสูตรอาหาร และมันยืดหยุ่นเพราะขึ้นอยู่กับข้อมูลนำเข้าจากผู้ใช้ ทั้งจำนวนสูตรอาหารและส่วนประกอบที่ใช้


    สำหรับฟังก์ชันนี้ เราสามารถลองแก้ไขทุกอย่างในพรอมต์เดียว หรือแบ่งเป็นสองพรอมต์ มาลองวิธีหลังกัน ที่นี่เราแนะนำการเพิ่มพรอมต์เพิ่ม แต่เพื่อให้ทำงานได้ เราต้องเพิ่มผลลัพธ์ของพรอมต์แรกเป็นบริบทให้กับพรอมต์ที่สอง

    ค้นหาส่วนในโค้ดที่พิมพ์ผลลัพธ์จากพรอมต์แรกและเพิ่มโค้ดต่อไปนี้ลงไปข้างล่าง:

    ```python
    old_prompt_result = response.output_text

    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    ```
