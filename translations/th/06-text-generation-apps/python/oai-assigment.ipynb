ถ้าคุณอยากเพิ่มประสบการณ์แบบนี้ลงในแอปของคุณเอง คุณต้องเข้าใจแนวคิดอย่าง prompt, completion และเลือกไลบรารีที่จะใช้ ซึ่งทั้งหมดนี้คุณจะได้เรียนรู้ในบทนี้

## แอปสร้างข้อความคืออะไร?

ปกติแล้วเวลาคุณสร้างแอป มันจะมีอินเทอร์เฟซแบบใดแบบหนึ่ง เช่น:

- แบบใช้คำสั่ง แอปคอนโซลเป็นตัวอย่างที่ดีที่คุณพิมพ์คำสั่งแล้วมันก็ทำงาน เช่น `git` เป็นแอปแบบใช้คำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) บางแอปมีอินเทอร์เฟซแบบกราฟิก (GUI) ที่คุณคลิกปุ่ม, พิมพ์ข้อความ, เลือกตัวเลือก ฯลฯ

### แอปคอนโซลและ UI มีข้อจำกัด

ลองเปรียบเทียบกับแอปแบบใช้คำสั่งที่คุณต้องพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ ต้องเป็นคำสั่งที่แอปรองรับเท่านั้น
- **จำกัดภาษาที่ใช้** บางแอปรองรับหลายภาษา แต่โดยปกติแล้วแอปจะถูกสร้างมาสำหรับภาษาใดภาษาหนึ่ง แม้ว่าจะเพิ่มการรองรับภาษาอื่นได้ก็ตาม

### ข้อดีของแอปสร้างข้อความ

แล้วแอปสร้างข้อความต่างจากนี้อย่างไร?

ในแอปสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น ไม่ได้จำกัดแค่ชุดคำสั่งหรือภาษาที่ป้อนเข้าไป คุณสามารถใช้ภาษาธรรมชาติในการโต้ตอบกับแอปได้ อีกข้อดีคือคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ถูกฝึกมาจากข้อมูลจำนวนมหาศาล ในขณะที่แอปแบบเดิมอาจจำกัดแค่ข้อมูลในฐานข้อมูลเท่านั้น

### แอปสร้างข้อความเอาไปทำอะไรได้บ้าง?

มีหลายอย่างที่คุณสามารถสร้างได้ เช่น:

- **แชทบอท** แชทบอทที่ตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและสินค้า ก็เป็นตัวอย่างที่ดี
- **ผู้ช่วย** LLM เก่งในเรื่องอย่างการสรุปข้อความ, หาข้อมูลเชิงลึกจากข้อความ, สร้างข้อความเช่นเรซูเม่ ฯลฯ
- **ผู้ช่วยเขียนโค้ด** ขึ้นอยู่กับโมเดลภาษา คุณสามารถสร้างผู้ช่วยเขียนโค้ดที่ช่วยคุณเขียนโค้ดได้ เช่น ใช้ GitHub Copilot หรือ ChatGPT เพื่อช่วยเขียนโค้ด

### สร้าง resource และหาคีย์ API ของคุณ

### ตั้งค่าการใช้งาน OpenAI

ถ้าคุณใช้ OpenAI ให้ตั้งค่าตามนี้:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"
```

ข้างบนนี้เราตั้งค่า:

- `api_key` คือ API key ของคุณที่ได้จาก OpenAI dashboard
- `deployment` คือเวอร์ชัน GPT ที่คุณใช้

> [!NOTE]
> `os.environ` เป็นฟังก์ชันสำหรับอ่าน environment variable คุณสามารถใช้เพื่ออ่าน environment variable อย่าง `OPENAI_API_KEY`

## สร้างข้อความ

วิธีสร้างข้อความคือใช้คลาส `chat.completion` ตัวอย่างเช่น:

```python
prompt = "Complete the following: Once upon a time there was a"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": prompt}])
print(completion.choices[0].message.content)
```

ในโค้ดข้างบน เราสร้าง completion object แล้วส่งโมเดลที่ต้องการใช้กับ prompt จากนั้นพิมพ์ข้อความที่สร้างออกมา

### Chat completions

ที่ผ่านมา คุณเห็นว่าเราใช้ `Completion` เพื่อสร้างข้อความ แต่ยังมีคลาสอีกตัวชื่อ `ChatCompletion` ที่เหมาะกับแชทบอทมากกว่า ตัวอย่างการใช้งาน:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-3.5-turbo"

completion = client.chat.completions.create(model=deployment, messages=[{"role": "user", "content": "Hello world"}])
print(completion.choices[0].message.content)
```

จะมีรายละเอียดเกี่ยวกับฟีเจอร์นี้ในบทถัดไป

## แบบฝึกหัด - สร้างแอปสร้างข้อความแรกของคุณ

ตอนนี้เราได้เรียนรู้วิธีตั้งค่าและคอนฟิก OpenAI service แล้ว ถึงเวลาสร้างแอปสร้างข้อความแรกของคุณ ทำตามขั้นตอนเหล่านี้เพื่อสร้างแอปของคุณ:


1. สร้าง virtual environment และติดตั้ง openai:

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นหากคุณใช้งานโน้ตบุ๊กนี้บน Codespaces หรือใน Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> ถ้าคุณใช้ Windows ให้พิมพ์ `venv\Scripts\activate` แทนที่จะใช้ `source venv/bin/activate`

> [!NOTE]
> คุณสามารถหา OpenAI key ของคุณได้โดยไปที่ https://platform.openai.com/settings/organization/api-keys แล้วค้นหาคำว่า `API keys` คุณสามารถสร้าง key ใหม่ที่นั่นและคัดลอกค่าได้ทันที


1. สร้างไฟล์ *app.py* และใส่โค้ดต่อไปนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages)

# print response
print(completion.choices[0].message.content)
````

```output
     very unhappy _____.

    Once upon a time there was a very unhappy mermaid.
    ```


## ประเภทของพรอมต์ที่แตกต่างกัน สำหรับงานที่หลากหลาย

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความด้วยพรอมต์แล้ว คุณยังมีโปรแกรมที่สามารถปรับแต่งและเปลี่ยนแปลงเพื่อสร้างข้อความในรูปแบบต่าง ๆ ได้อีกด้วย

พรอมต์สามารถนำไปใช้กับงานได้หลากหลาย เช่น:

- **สร้างข้อความในรูปแบบต่าง ๆ** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ ฯลฯ
- **ค้นหาข้อมูล** คุณสามารถใช้พรอมต์เพื่อค้นหาข้อมูล เช่น ตัวอย่างนี้ 'CORS หมายถึงอะไรในงานพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น การพัฒนา regular expression สำหรับตรวจสอบอีเมล หรือแม้แต่สร้างโปรแกรมทั้งโปรแกรม เช่น เว็บแอปพลิเคชัน

## แบบฝึกหัด - สร้างตัวสร้างสูตรอาหาร

ตอนนี้เราได้ลองเล่นสถานการณ์ตัวอย่างไปแล้ว มาลองเขียนโค้ดให้ตรงกับสถานการณ์ที่แสดงไว้กันเถอะ ทำตามขั้นตอนดังนี้:

1. ใช้ไฟล์ *app.py* ที่มีอยู่แล้วเป็นจุดเริ่มต้น
1. หา `prompt` ตัวแปร แล้วเปลี่ยนโค้ดของมันเป็นดังนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

> NOTE, โมเดล LLM ของคุณไม่แน่นอน ดังนั้นคุณอาจได้ผลลัพธ์ที่ต่างกันทุกครั้งที่รันโปรแกรม

เยี่ยมเลย มาดูกันว่าเราจะปรับปรุงอะไรได้บ้าง เพื่อให้ดีขึ้น เราต้องการให้โค้ดยืดหยุ่นมากขึ้น เพื่อจะได้ปรับเปลี่ยนวัตถุดิบและจำนวนสูตรอาหารได้ตามต้องการ


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````

การทดสอบโค้ดอาจมีหน้าตาแบบนี้:

```output

List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract

เพื่อให้แอปดีขึ้น เราต้องการเพิ่มสิ่งต่อไปนี้:

- **กรองวัตถุดิบ** เราต้องการสามารถกรองวัตถุดิบที่เราไม่ชอบหรือแพ้ออกได้ เพื่อทำการเปลี่ยนแปลงนี้ เราสามารถแก้ไข prompt ที่มีอยู่และเพิ่มเงื่อนไขการกรองไว้ท้าย prompt ได้แบบนี้:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"

    จะเห็นได้ว่าสูตรอาหารที่มีนมจะถูกกรองออก แต่ถ้าคุณแพ้แลคโตส คุณอาจอยากกรองสูตรที่มีชีสด้วย ดังนั้นควรระบุให้ชัดเจน

    ```python
    
- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = completion.choices[0].text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
    
    # แสดงผลลัพธ์
    print("Shopping list:")
    print(completion.choices[0].message.content)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_tokens` is 1200. 

        ```python
        completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=1200)
        ```

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, max_tokens=100)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    completion = client.chat.completion.create(model=deployment, messages=messages, temperature=0.5)
