สำหรับคุณที่จะเพิ่มประสบการณ์เช่นนั้นในแอป คุณต้องเข้าใจแนวคิดเช่นพรอมต์, การเติมข้อความ และเลือกไลบรารีที่จะใช้ นั่นคือสิ่งที่คุณจะได้เรียนรู้ในบทนี้

## เป้าหมายการเรียนรู้

เมื่อจบบทเรียนนี้ คุณจะสามารถ:

## แอปสร้างข้อความคืออะไร?

ปกติแล้วเมื่อคุณสร้างแอป จะมีอินเทอร์เฟซบางชนิด เช่น:

- แบบคำสั่ง แอปคอนโซลเป็นแอปประเภทที่คุณพิมพ์คำสั่งและมันก็ทำงาน ตัวอย่างเช่น `git` คือแอปแบบคำสั่ง
- อินเทอร์เฟซผู้ใช้ (UI) บางแอปมีอินเทอร์เฟซกราฟิก (GUI) ที่คุณคลิกปุ่ม, ป้อนข้อความ, เลือกตัวเลือก และอื่น ๆ

### แอปคอนโซลและ UI มีข้อจำกัด

เปรียบเทียบกับแอปแบบคำสั่งที่คุณพิมพ์คำสั่ง:

- **มีข้อจำกัด** คุณไม่สามารถพิมพ์คำสั่งอะไรก็ได้ มีแค่คำสั่งที่แอปรองรับเท่านั้น
- **เฉพาะเจาะจงภาษา** บางแอปรองรับหลายภาษา แต่โดยค่าเริ่มต้นแอปถูกสร้างสำหรับภาษาหนึ่ง แม้ว่าคุณสามารถเพิ่มการรองรับภาษาได้

### ประโยชน์ของแอปสร้างข้อความ

แล้วแอปสร้างข้อความต่างอย่างไร?

ในแอปสร้างข้อความ คุณมีความยืดหยุ่นมากขึ้น ไม่จำกัดให้เฉพาะชุดคำสั่งหรือภาษาป้อนข้อมูลเฉพาะ คุณสามารถใช้ภาษาธรรมชาติในการโต้ตอบกับแอป อีกประโยชน์คือเพราะคุณกำลังโต้ตอบกับแหล่งข้อมูลที่ผ่านการฝึกอบรมจากข้อมูลจำนวนมาก ในขณะที่แอปแบบดั้งเดิมอาจถูกจำกัดอยู่แค่ฐานข้อมูลที่มี

### ฉันสามารถสร้างอะไรได้บ้างด้วยแอปสร้างข้อความ?

มีหลายสิ่งที่คุณสามารถสร้างได้ เช่น:

- **แชทบอท** แชทบอทตอบคำถามเกี่ยวกับหัวข้อต่าง ๆ เช่น บริษัทของคุณและผลิตภัณฑ์ อาจเหมาะกับคุณ
- **ผู้ช่วย** LLM ดีเยี่ยมสำหรับการสรุปข้อความ, วิเคราะห์ข้อมูลจากข้อความ, สร้างข้อความเช่นเรซูเม่ และอื่น ๆ
- **ผู้ช่วยโค้ด** ขึ้นกับโมเดลภาษาที่คุณใช้ คุณสามารถสร้างผู้ช่วยเขียนโค้ด ตัวอย่างเช่น ใช้ผลิตภัณฑ์เช่น GitHub Copilot รวมถึง ChatGPT เพื่อช่วยเขียนโค้ด

## แอปแรกโดยใช้ openai

มาดูวิธีสร้างแอปแรกของเรา ไลบรารีที่ต้องใช้ จำนวนที่ต้องการ และอื่น ๆ

### ติดตั้ง openai

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นถ้าคุณรันสมุดบันทึกนี้ใน Codespaces หรือใน Devcontainer


มีไลบรารีมากมายสำหรับโต้ตอบกับ OpenAI หรือ Azure OpenAI คุณสามารถใช้หลายภาษาโปรแกรม เช่น C#, Python, JavaScript, Java และอื่น ๆ  
เราเลือกใช้ไลบรารี `openai` สำหรับ Python ดังนั้นเราจะใช้ `pip` ในการติดตั้ง

```bash
pip install openai
```

ถ้าคุณไม่ได้รันสมุดบันทึกนี้ใน Codespaces หรือ Dev Container คุณต้องติดตั้ง [Python](https://www.python.org/) บนเครื่องของคุณด้วย

### สร้างทรัพยากรและหาคีย์ API ของคุณ

ในกรณีที่คุณยังไม่ได้ดำเนินการ คุณต้องทำตามขั้นตอนดังนี้:

- สร้างบัญชีที่ OpenAI <https://platform.openai.com/signup>
- จากนั้นรับคีย์ API ของคุณที่ <https://platform.openai.com/api-keys>

>[!NOTE]
> ควรแยกคีย์ API ของคุณออกจากโค้ด คุณทำได้โดยใช้ตัวแปรสภาพแวดล้อม
> - กำหนดค่าตัวแปรสภาพแวดล้อม `OPENAI_KEY` เป็นคีย์ API ของคุณในไฟล์ .env ถ้าคุณทำแบบฝึกหัดก่อนหน้าในหลักสูตรนี้แล้ว ทุกอย่างก็พร้อมแล้ว
> - สิ่งสำคัญคือคีย์ API จะเข้าถึงได้เพียงครั้งเดียว ดังนั้นจึงสำคัญมากที่จะต้องตรวจสอบว่าคุณได้คัดลอกถูกต้อง หากมันไม่ทำงานตามที่คาดไว้ ขอแนะนำให้ลบคีย์แล้วสร้างคีย์ใหม่


### ตั้งค่าการกำหนดค่า OpenAI

ถ้าคุณใช้ OpenAI นี่คือวิธีตั้งค่าการกำหนดค่า:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

ด้านบนเรากำหนดค่า:

- `api_key` นี่คือคีย์ API ของคุณที่พบในแดชบอร์ด OpenAI
- `deployment` นี่คือเวอร์ชัน GPT ของคุณ

> [!NOTE]
> `os.environ` เป็นฟังก์ชันที่อ่านตัวแปรสภาพแวดล้อม คุณสามารถใช้มันเพื่ออ่านตัวแปรสภาพแวดล้อมเช่น `OPENAI_API_KEY`

## สร้างข้อความ

วิธีสร้างข้อความคือใช้เมธอด `responses.create` นี่คือตัวอย่าง:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

ในโค้ดข้างต้น เราสร้างออบเจ็กต์การตอบกลับและส่งโมเดลที่เราต้องการใช้และพรอมต์เข้าไป จากนั้นพิมพ์ข้อความที่สร้างขึ้น

### การตอบกลับแชท

จนถึงตอนนี้ คุณได้เห็นว่าเราสามารถสร้างข้อความจากพรอมต์เดียวอย่างไร API Responses ยังเหมาะกับแชทบอทด้วย – คุณส่งรายการข้อความเป็น `input` เพื่อสร้างการสนทนา นี่คือตัวอย่างการใช้งาน:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

จะมีการอธิบายฟังก์ชันนี้เพิ่มเติมในบทต่อ ๆ ไป

## แบบฝึกหัด – แอปสร้างข้อความแรกของคุณ

ตอนนี้ที่เราเรียนรู้วิธีตั้งค่าและกำหนดค่า OpenAI แล้ว ก็ถึงเวลาสร้างแอปสร้างข้อความแรกของคุณ ทำตามขั้นตอนเหล่านี้เพื่อสร้างแอปของคุณ:


1. สร้างสภาพแวดล้อมเสมือนและติดตั้ง openai:

  > [!NOTE] ขั้นตอนนี้ไม่จำเป็นถ้าคุณรันโน้ตบุ๊กนี้บน Codespaces หรือภายใน Devcontainer


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> หากคุณใช้ Windows ให้พิมพ์ `venv\Scripts\activate` แทน `source venv/bin/activate`

> [!NOTE]
> ค้นหากุญแจ OpenAI ของคุณโดยไปที่ https://platform.openai.com/settings/organization/api-keys แล้วค้นหา `API keys` คุณสามารถสร้างกุญแจใหม่ที่นั่นและคัดลอกค่าได้ทันที


1. สร้างไฟล์ *app.py* และใส่โค้ดดังต่อไปนี้:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    คุณควรเห็นผลลัพธ์ประมาณดังนี้:

    ```output
     very unhappy _____.

    ครั้งหนึ่งนานมาแล้ว มีนางเงือกที่ทุกข์ใจมาก
    ```


## ประเภทของพรอมต์ที่แตกต่างกัน สำหรับสิ่งต่าง ๆ

ตอนนี้คุณได้เห็นวิธีการสร้างข้อความโดยใช้พรอมต์แล้ว คุณยังมีโปรแกรมที่ใช้งานได้ซึ่งคุณสามารถแก้ไขและเปลี่ยนแปลงเพื่อสร้างข้อความประเภทต่าง ๆ

พรอมต์สามารถใช้สำหรับงานหลากหลายประเภท ตัวอย่างเช่น:

- **สร้างข้อความประเภทหนึ่ง** เช่น คุณสามารถสร้างบทกวี คำถามสำหรับแบบทดสอบ เป็นต้น
- **ค้นหาข้อมูล** คุณสามารถใช้พรอมต์เพื่อค้นหาข้อมูล เช่น ตัวอย่างต่อไปนี้ 'CORS หมายถึงอะไรในการพัฒนาเว็บ?'
- **สร้างโค้ด** คุณสามารถใช้พรอมต์เพื่อสร้างโค้ด เช่น การพัฒนากฎการตรวจสอบ (regular expression) เพื่อใช้ตรวจสอบอีเมล หรือต้องการสร้างโปรแกรมทั้งโปรแกรม เช่น เว็บแอปได้เลย

## แบบฝึกหัด - สร้างเครื่องมือสร้างสูตรอาหาร  

ตอนนี้ที่เราได้เล่นสถานการณ์แล้ว มาเขียนโค้ดให้ตรงกับสถานการณ์ที่สาธิตกันเถอะ เพื่อทำเช่นนั้น ให้ทำตามขั้นตอนดังนี้:  

1. ใช้ไฟล์ *app.py* ที่มีอยู่เป็นจุดเริ่มต้น  
1. ค้นหาตัวแปร `prompt` และเปลี่ยนโค้ดของมันเป็นดังนี้:  


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

ดีมาก มาดูกันว่าทำอย่างไรเราจึงจะปรับปรุงสิ่งต่าง ๆ ได้ ในการปรับปรุง เราต้องการให้โค้ดยืดหยุ่น เพื่อให้ส่วนผสมและจำนวนสูตรอาหารสามารถพัฒนาและเปลี่ยนแปลงได้

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

    old_prompt_result = response.output_text
