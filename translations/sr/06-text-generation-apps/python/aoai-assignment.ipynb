## Шта је апликација за генерисање текста?

Обично када правите апликацију, она има неки вид интерфејса као што је следеће:

- На командној линији. Конзолне апликације су типичне апликације где куцате команду и она извршава задатак. На пример, `git` је апликација заснована на командној линији.
- Кориснички интерфејс (UI). Неке апликације имају графички кориснички интерфејс (GUI) где кликћете на дугмад, уносите текст, бирајте опције и више.

### Конзолне и UI апликације имају ограничења

Упоредите то са апликацијом заснованом на командној линији где куцате команду:

- **Ограничено је**. Не можете само куцати било коју команду, већ само оне које апликација подржава.
- **Језички специфично**. Неке апликације подржавају много језика, али по дефаулту апликација је направљена за одређени језик, иако можете додати подршку за више језика.

### Предности апликација за генерисање текста

Па како се апликација за генерисање текста разликује?

У апликацији за генерисање текста имате више флексибилности, нисте ограничени на скуп команди или одређени језик уноса. Уместо тога, можете користити природни језик за интеракцију са апликацијом. Друга предност је што већ комуницирате са извором података који је обучен на огромном корпусу информација, док је традиционална апликација можда ограничена на оно што се налази у бази података.

### Шта могу да направим са апликацијом за генерисање текста?

Постоји много ствари које можете направити. На пример:

- **Чатбот**. Чатбот који одговара на питања о темама као што су ваша компанија и њени производи може бити добар избор.
- **Помоћник**. LLM-ови су одлични за ствари попут резимирања текста, добијања увида из текста, производње текста као што су резимеа и више.
- **Асистент за кодирање**. У зависности од модела језика који користите, можете направити асистента за кодирање који вам помаже да пишете код. На пример, можете користити производ као што је GitHub Copilot као и ChatGPT да вам помогну у писању кода.

### Поставите конфигурацију за Azure

Ако користите Azure OpenAI, ево како да подесите конфигурацију. Responses API се сервира са Azure OpenAI **v1 endpoint-а**, па усмеравамо `OpenAI` клијента ка `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Горенаведено поставља следеће:

- `api_key`, ово је ваш API кључ пронађен у Azure порталу.
- `base_url`, ово је ваш Azure OpenAI endpoint са додатком `/openai/v1/`. Endpoint можете наћи у Azure порталу поред вашег API кључа. Коришћење v1 endpoint-а значи да више није потребно прослеђивати `api_version`.
- `deployment`, ово је име распореда модела који сте креирали у Foundry порталу.

> [!NOTE]
> `os.environ` је мапа која чита променљиве окружења. Можете је користити да прочитате променљиве окружења као `AZURE_OPENAI_API_KEY` и `AZURE_OPENAI_ENDPOINT`.

## Генеришите текст

Начин за генерисање текста је коришћење метода `responses.create`. Ево примера:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

У горњем коду, ми креирамо response објекат и преносимо модел који желимо да користимо и промпт. Онда штампамо генерисани текст.

### Одговори у ћаскању

Responses API је одличан и за једноокретно генерисање текста и за вишетокене ћаскање – само дајте више порука у списку `input` да бисте изградили разговор:

```python

```

## Вежба - направи генератор рецепата

Сада када смо одиграли сценарио, хајде да напишемо код који одговара приказаном сценарију. Да бисмо то урадили, пратите следеће кораке:

1. Користите постојећу датотеку *app.py* као почетну тачку
1. Пронађите променљиву `prompt` и промените њен код у следеће:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > Note, the closer to 1.0, the more varied the output.
