## Шта је апликација за генерисање текста?

Обично када креирате апликацију има неки вид интерфејса као што је следеће:

- На бази команди. Конзолне апликације су типичне апликације где куцате команду и она извршава задатак. На пример, `git` је апликација на бази команди.
- Кориснички интерфејс (UI). Неке апликације имају графички кориснички интерфејс (GUI) где кликћете на дугмад, уносите текст, одабирате опције и више.

### Конзолне и UI апликације су ограничене

Упоредите са апликацијом на бази команди где куцате команду:

- **Ограничена је**. Не можете једноставно куцати било коју команду, већ само оне које апликација подржава.
- **Специфична за језик**. Неке апликације подржавају више језика, али по подразумеваној вредности апликација је направљена за одређени језик, чак и ако можете додати подршку за више језика.

### Предности апликација за генерисање текста

Како се онда разликује апликација за генерисање текста?

У апликацији за генерисање текста имате већу флексибилност, нисте ограничени сетом команди или одређеним улазним језиком. Уместо тога, можете користити природни језик за интеракцију са апликацијом. Још једна предност је што већ комуницирате са извором података који је трениран на огромном корпусу информација, док традиционална апликација може бити ограничена на оно што је у бази података.

### Шта могу да направим са апликацијом за генерисање текста?

Постоје многе ствари које можете направити. На пример:

- **Чет-бот**. Чет-бот који одговара на питања о темама, као што су ваша компанија и њени производи, може бити добар избор.
- **Помоћник**. Вештачки језички модели сјајни су у стварима као што су резимирање текста, добијање увида из текста, стварање текста као што су биографије и више.
- **Асистент за кодирање**. У зависности од модела језика који користите, можете направити асистента за кодирање који вам помаже да пишете код. На пример, можете користити производе као што су GitHub Copilot и ChatGPT за помоћ при писању кода.

### Конфигурисање Azure

Ако користите Azure OpenAI, ево како се конфигурише. Responses API се нуди са Azure OpenAI **v1 endpoint-а**, па показујемо `OpenAI` клијенту на `<your-endpoint>/openai/v1/`:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

Горe подешавамо следеће:

- `api_key`, ово је ваш API кључ који сте добили у Azure Порталу.
- `base_url`, ово је ваш Azure OpenAI endpoint са додатком `/openai/v1/`. Endpoint можете пронаћи у Azure Порталу поред вашег API кључа. Користећи v1 endpoint више није потребно пренети `api_version`.
- `deployment`, ово је име распореда модела који сте креирали у Foundry порталу.

> [!NOTE]
> `os.environ` је мапа која чита променљиве окружења. Можете је користити за читање променљивих окружења као што су `AZURE_OPENAI_API_KEY` и `AZURE_OPENAI_ENDPOINT`.

## Генерисање текста

Начин за генерисање текста је коришћење методе `responses.create`. Ево примера:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

У горе наведеном коду креирамо објекат за одговор и прослеђујемо модел који желимо да користимо и упит. Затим штампамо генерисани текст.

### Чет одговори

Responses API је добро прилагођен како за једнокружно генерисање текста тако и за вишекружне чет-ботове – једноставно пружате више порука у листи `input` да бисте изграђивали разговор:

```python

```

## Вежба - направите генератор рецепата

Сада када смо одиграли сценарио, хајде да напишемо код који одговара приказаном сценарију. Да бисте то урадили, пратите ове кораке:

1. Користите постојећу датотеку *app.py* као полазну тачку
1. Пронађите променљиву `prompt` и промените њен код на следећи начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Note, the closer to 1.0, the more varied the output.
