## Шта је апликација за генерисање текста?

Обично, када правите апликацију, она има неку врсту интерфејса као што је следеће:

- Командно заснован. Конзолне апликације су типичне апликације где уписујете команду и она обавља задатак. На пример, `git` је командна апликација.
- Кориснички интерфејс (UI). Неке апликације имају графичке корисничке интерфејсе (GUI) где кликате на дугмад, уносите текст, бирани опције и још много тога.

### Конзолне и UI апликације имају ограничења

Упоредимо са командном апликацијом где уписујете команду:

- **Ограничавајe**. Не можете уписати било коју команду, већ само оне које апликација подржава.
- **Језички специфично**. Неке апликације подржавају много језика, али по дифолту су направљене за одређени језик, иако се може додати подршка за друге језике.

### Предности апликација за генерисање текста

Како се онда апликација за генерисање текста разликује?

У апликацији за генерисање текста имате већу флексибилност, нисте ограничени сетом команди или специфичним улазним језиком. Уместо тога, можете користити природни језик за интеракцију са апликацијом. Друга предност је што већ комуницирате са извором података који је обучен на огромном корпусу информација, док би традиционална апликација била ограничена на оно што има у бази података.

### Шта могу да направим користећи апликацију за генерисање текста?

Много тога се може направити. На пример:

- **Чатбот**. Чатбот који одговара на питања о темама као што су ваша компанија и њени производи може бити добра опција.
- **Помоћник**. Велику моћ имају LLM-ови за ствари попут резимирања текста, добијања увида из текста, прављења текста као што су биографије и још много тога.
- **Асистент за кодирање**. У зависности од модела језика који користите, можете направити асистента за кодирање који вам помаже да пишете код. На пример, можете користити производе као што су GitHub Copilot као и ChatGPT за помоћ у писању кода.

## Генерисање текста

Начин да генеришете текст је коришћење методе `responses.create`. Ево примера:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

У горе наведеном коду, правимо објекат одговора и прослеђујемо модел који желимо да користимо и упућивање. Затим штампамо генерисани текст.

### Одговори у ћаскању

До сада сте видели како генерисати текст из једног упућивања. Responses API је такође погодан за чатботове - унесете листу порука као `input` да бисте саставили разговор. Ево примера коришћења:


## Вежба - направите генератор рецепата

Сада када смо одиграли сценарио, хајде да напишемо код који одговара демонстрираном сценарију. Да бисте то урадили, пратите ове кораке:

1. Користите постојећи *app.py* фајл као почетну тачку
1. Пронађите променљиву `prompt` и измените њен код у следеће:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> НАПОМЕНА, ваш LLM је недетерминистички, па је могуће да добијете различите резултате сваки пут када покренете програм.

Одлично, хајде да видимо како можемо да унапредимо ствари. Да бисмо унапредили ствари, желимо да обезбедимо флексибилност кода, тако да састојци и број рецепата могу бити побољшани и промењени.


1. Хајде да променимо код на следећи начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
