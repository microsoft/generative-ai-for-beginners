## Шта је апликација за генерисање текста?

Обично када правите апликацију, она има неки вид интерфејса као што је следеће:

- Командна линија. Апликације на командној линији су типичне апликације где укуцате команду и она обави задатак. На пример, `git` је апликација заснована на командној линији.
- Кориснички интерфејс (UI). Неке апликације имају графичке корисничке интерфејсе (GUI) где кликате на тастере, уносите текст, бира и друге опције.

### Апликације на командној линији и UI су ограничене

Упоредите то са апликацијом заснованом на командној линији где укуцате команду:

- **Ограничено је**. Не можете укуцати било коју команду, само оне које апликација подржава.
- **Језички специфично**. Неке апликације подржавају више језика, али подразумевано је направљена за одређени језик, иако можете додати подршку за друге језике.

### Предности апликација за генерисање текста

Како се онда апликација за генерисање текста разликује?

У апликацији за генерисање текста имате већу флексибилност, нисте ограничени на скуп команди или на одређени улазни језик. Уместо тога, можете користити природни језик за интеракцију са апликацијом. Друга предност је што већ комуницирате са извором података који је обучен на огромном корпусу информација, док традиционална апликација може бити ограничена на садржај базе података.

### Шта могу направити са апликацијом за генерисање текста?

Постоји много ствари које можете направити. На пример:

- **Чатбот**. Чатбот који одговара на питања о темама, као што је ваша компанија и њени производи, могао би бити добар избор.
- **Помоћник**. Велики језички модели су одлични у ствари као што су сажимање текста, добијање увида из текста, креирање текста као што су резимеа и друго.
- **Асистент за кодирање**. У зависности од језичког модела који користите, можете направити асистента за кодирање који вам помаже да пишете код. На пример, можете користити производ као што је GitHub Copilot као и ChatGPT да вам помогне у писању кода.

## Генерисање текста

Начин за генерисање текста је коришћење методе `responses.create`. Ево примера:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

У горе наведеном коду, правимо објекат одговора и прослеђујемо модел који желимо да користимо као и упит. Онда штампамо генерисани текст.

### Одговори у чату

До сада сте видели како можемо да генеришемо текст из једног упита. Responses API је такође веома погодан за чатботове – прослеђујете листу порука као `input` да бисте изграђивали разговор. Ево примера коришћења:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## Вежба - направите генератор рецепата

Сада када смо одиграли сценарио, хајде да напишемо код који одговара демонстрираном сценарију. Да бисте то урадили, пратите следеће кораке:

1. Користите постојећи *app.py* фајл као полазну тачку
1. Пронађите променљиву `prompt` и промените њен код на следећи начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> НАПОМЕНА, ваш LLM је недетерминистички, тако да можете добити различите резултате сваки пут када покренете програм.

Одлично, хајде да видимо како можемо да унапредимо ствари. Да бисмо унапредили ствари, желимо да код буде флексибилан, тако да састојци и број рецепата могу да се побољшају и промене.


1. Хајде да променимо код на следећи начин:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Побољшање додавањем филтера и листе за куповину
