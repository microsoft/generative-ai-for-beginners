## Hvad er en tekstgenereringsapp?

Normalt når du bygger en app, har den en slags brugerflade som følgende:

- Kommandobaseret. Konsolapps er typiske apps, hvor du skriver en kommando, og den udfører en opgave. For eksempel er `git` en kommandobaseret app.
- Brugerflade (UI). Nogle apps har grafiske brugerflader (GUIs), hvor du klikker på knapper, indtaster tekst, vælger muligheder og mere.

### Konsol- og UI-apps er begrænsede

Sammenlign det med en kommandobaseret app, hvor du skriver en kommando:

- **Den er begrænset**. Du kan ikke bare skrive en hvilken som helst kommando, kun dem appen understøtter.
- **Sprogafhængig**. Nogle apps understøtter mange sprog, men som standard er appen bygget til et specifikt sprog, selvom du kan tilføje mere sprogsupport.

### Fordele ved tekstgenereringsapps

Hvordan adskiller en tekstgenereringsapp sig så?

I en tekstgenereringsapp har du mere fleksibilitet, du er ikke begrænset til et sæt kommandoer eller et specifikt input-sprog. I stedet kan du bruge naturligt sprog til at interagere med appen. En anden fordel er, at fordi du allerede interagerer med en datakilde, der er trænet på en enorm mængde information, hvor en traditionel app måske er begrænset af det, der findes i en database.

### Hvad kan jeg bygge med en tekstgenereringsapp?

Der er mange ting, du kan bygge. For eksempel:

- **En chatbot**. En chatbot, der besvarer spørgsmål om emner som din virksomhed og dens produkter, kan være et godt match.
- **Hjælper**. Sprogmodeller (LLMs) er gode til ting som at opsummere tekst, få indsigt fra tekst, producere tekst som CV'er og mere.
- **Kodeassistent**. Afhængigt af den sproglige model, du bruger, kan du bygge en kodeassistent, der hjælper dig med at skrive kode. For eksempel kan du bruge produkter som GitHub Copilot samt ChatGPT til at hjælpe dig med at skrive kode.

## Generer tekst

Måden at generere tekst på er at bruge metoden `responses.create`. Her er et eksempel:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

I koden ovenfor opretter vi et svarobjekt og sender den model, vi vil bruge, samt prompten. Derefter printer vi den genererede tekst.

### Chat-svar

Indtil videre har du set, hvordan vi kan generere tekst ud fra en enkelt prompt. Responses API'et er også velegnet til chatbots - du giver en liste af beskeder som `input` for at opbygge en samtale. Her er et eksempel på at bruge det:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## Øvelse - byg en opskriftsgenerator

Nu hvor vi har spillet et scenarie igennem, lad os skrive kode, der matcher det viste scenarie. For at gøre det, følg disse trin:

1. Brug den eksisterende *app.py* fil som udgangspunkt
1. Find variablen `prompt` og ændr dens kode til følgende:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Fremragende, lad os se, hvordan vi kan forbedre tingene. For at forbedre tingene vil vi sikre, at koden er fleksibel, så ingredienser og antal opskrifter kan forbedres og ændres.


1. Lad os ændre koden på følgende måde:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Forbedr ved at tilføje filter og indkøbsliste

Vi har nu en fungerende app i stand til at producere opskrifter, og den er fleksibel, da den bygger på input fra brugeren, både i forhold til antal opskrifter men også de brugte ingredienser.

For at forbedre den yderligere vil vi tilføje følgende:

- **Filtrer ingredienser fra**. Vi vil kunne filtrere ingredienser fra, som vi ikke kan lide eller er allergiske over for. For at opnå denne ændring kan vi redigere vores eksisterende prompt og tilføje en filterbetingelse til slutningen af den som følger:

    ```python
    filter = input("Filter (for example, vegetarian, vegan, or gluten-free: ")

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"
    ```

    Ovenfor tilføjer vi `{filter}` til slutningen af prompten, og vi fanger også filterværdien fra brugeren.

- **Lav en indkøbsliste**. Vi vil lave en indkøbsliste, hvor vi tager højde for, hvad vi allerede har derhjemme.

    Til denne funktionalitet kan vi enten prøve at løse det hele i én prompt eller opdele det i to prompts. Lad os prøve sidstnævnte metode. Her foreslår vi at tilføje en ekstra prompt, men for at det skal fungere, skal vi føje resultatet af den første prompt til konteksten af den anden prompt.

    Find det sted i koden, hvor resultatet af den første prompt printes, og tilføj følgende kode nedenunder:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # udskriv svar
    print("Shopping list:")
    print(response.output_text)
    ```

    Bemærk følgende:

    - Vi opbygger en ny prompt ved at tilføje resultatet af den første prompt til den nye prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - Vi laver en ny forespørgsel, men tager også højde for antallet af tokens, vi bad om i den første prompt, så denne gang siger vi `max_output_tokens` er 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        Ved at prøve denne kode, kommer vi nu til følgende output:


- **Et ord om tokenlængde**. Vi bør overveje, hvor mange tokens vi skal bruge for at generere den ønskede tekst. Tokens koster penge, så hvor det er muligt, bør vi forsøge at være økonomiske med antallet af brugte tokens. Kan vi f.eks. formulere prompten, så vi kan bruge færre tokens?

   For at ændre brugte tokens kan du bruge parameteren `max_output_tokens`. For eksempel, hvis du vil bruge 100 tokens, ville du gøre:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Eksperimenter med temperatur**. Temperatur er noget, vi ikke har nævnt før, men det er et vigtigt element for, hvordan vores program opfører sig. Jo højere temperatur, jo mere tilfældigt bliver outputtet. Omvendt, jo lavere temperatur, jo mere forudsigeligt bliver outputtet. Overvej, om du ønsker variation i outputtet eller ej.

   For at ændre temperaturen kan du bruge parameteren `temperature`. For eksempel, hvis du vil bruge en temperatur på 0.5, ville du gøre:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > Bemærk, jo tættere på 1.0, jo mere varieret bliver outputtet.

- Byg en "studiekammerat". Denne app skal kunne svare på spørgsmål om et emne, for eksempel Python, du kunne have prompts som "Hvad er et bestemt emne i Python?", eller du kunne have en prompt, der siger, vis mig kode til et bestemt emne osv.
- Historiebot, giv historien liv, instruer botten til at spille en bestemt historisk person og spørg den om dets liv og tid.

## Løsning

### Studiekammerat

- "Du er ekspert i Python-sproget

    Foreslå en begynderlektion i Python i følgende format:
    
    Format:
    - begreber:
    - kort forklaring af lektionen:
    - øvelse i kode med løsninger"

Ovenfor er en startprompt, se hvordan du kan bruge den og justere den efter din smag.

### Historiebot

Her er nogle prompts, du kunne bruge:

- "Du er Abe Lincoln, fortæl mig om dig selv i 3 sætninger, og svar ved at bruge grammatik og ord, som Abe ville have brugt"
- "Du er Abe Lincoln, svar ved at bruge grammatik og ord, som Abe ville have brugt:

   Fortæl mig om dine største bedrifter, på 300 ord:"

## Videnscheck

Hvad gør begrebet temperatur?

1. Det kontrollerer, hvor tilfældigt outputtet er.
1. Det kontrollerer, hvor stort svaret er.
1. Det kontrollerer, hvor mange tokens der bruges.

A: 1

Hvad er en god måde at gemme hemmeligheder som API-nøgler på?

1. I kode.
1. I en fil.
1. I miljøvariable.

A: 3, fordi miljøvariable ikke gemmes i koden og kan læses fra koden.
