# 텍스트 생성 앱 빌드하기

지금까지 이 커리큘럼을 통해 프롬프트와 "프롬프트 엔지니어링"이라는 전체 분야와 같은 핵심 개념들을 보았습니다. ChatGPT, Office 365, Microsoft Power Platform 등과 같이 상호작용할 수 있는 많은 도구들이 어떤 작업을 수행하기 위해 프롬프트를 사용하는 것을 지원합니다.

이러한 경험을 앱에 추가하려면 프롬프트, 완료(completions)와 같은 개념을 이해하고 작업할 라이브러리를 선택해야 합니다. 바로 이 장에서 그 내용을 배웁니다.

## 소개

이 장에서 당신은:

- openai 라이브러리와 그 핵심 개념에 대해 배웁니다.
- openai를 사용하여 텍스트 생성 앱을 빌드합니다.
- 텍스트 생성 앱을 빌드하기 위해 프롬프트, 온도, 토큰과 같은 개념을 사용하는 방법을 이해합니다.

## 학습 목표

이 수업이 끝나면 당신은:

- 텍스트 생성 앱이 무엇인지 설명할 수 있습니다.
- openai를 사용하여 텍스트 생성 앱을 빌드할 수 있습니다.
- 더 많거나 적은 토큰을 사용하도록 앱을 구성하고 출력 결과를 다양하게 하기 위해 온도도 변경할 수 있습니다.

## 텍스트 생성 앱이란?

일반적으로 앱을 빌드할 때 다음과 같은 인터페이스가 있습니다:

- 명령 기반. 콘솔 앱은 명령어를 입력하면 작업을 수행하는 전형적인 앱입니다. 예를 들어, `git`은 명령 기반 앱입니다.
- 사용자 인터페이스(UI). 일부 앱에는 버튼을 클릭하고 텍스트를 입력하며 옵션을 선택하는 그래픽 사용자 인터페이스(GUI)가 있습니다.

### 콘솔 및 UI 앱의 한계

명령어를 입력하는 명령 기반 앱과 비교해 보세요:

- **한계가 있습니다**. 앱이 지원하는 명령어만 입력할 수 있고 아무 명령어나 입력할 수 없습니다.
- **언어 특정적입니다**. 일부 앱은 여러 언어를 지원하지만 기본적으로 특정 언어용으로 만들어져 있고 추가 언어 지원도 가능합니다.

### 텍스트 생성 앱의 장점

그렇다면 텍스트 생성 앱은 어떻게 다를까요?

텍스트 생성 앱에서는 더 많은 유연성이 있고 제한된 명령어나 특정 입력 언어에 얽매이지 않습니다. 대신 자연어를 사용하여 앱과 상호작용할 수 있습니다. 또 다른 장점은 기존 앱이 데이터베이스 내의 정보에 제한되는 반면, 텍스트 생성 앱은 방대한 코퍼스로 훈련된 데이터 소스와 이미 상호작용한다는 점입니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있나요?

만들 수 있는 것들이 많습니다. 예를 들어:

- <strong>챗봇</strong>. 회사 및 제품에 관한 질문에 답하는 챗봇도 좋은 예가 될 수 있습니다.
- <strong>도우미</strong>. LLM은 텍스트 요약, 텍스트 인사이트 추출, 이력서와 같은 텍스트 작성에 매우 뛰어납니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코드를 작성하는 데 도움을 주는 코드 어시스턴트를 만들 수 있습니다. 예를 들어, GitHub Copilot이나 ChatGPT를 사용할 수 있습니다.

## 어떻게 시작할 수 있나요?

보통 LLM과 통합하려면 다음 두 가지 접근법 중 하나를 택합니다:

- API를 사용합니다. 프롬프트를 포함한 웹 요청을 구성하고 생성된 텍스트를 받아옵니다.
- 라이브러리를 사용합니다. 라이브러리는 API 호출을 캡슐화하여 사용하기 쉽게 만듭니다.

## 라이브러리/SDK

LLM과 작업할 때 잘 알려진 몇 가지 라이브러리가 있습니다:

- **openai**. 이 라이브러리는 모델에 쉽게 연결하고 프롬프트를 보낼 수 있게 해줍니다.

그 외 상위 수준에서 작동하는 라이브러리도 있습니다:

- **Langchain**. Langchain은 잘 알려져 있으며 Python을 지원합니다.
- **Semantic Kernel**. Semantic Kernel은 Microsoft의 라이브러리로 C#, Python, Java를 지원합니다.

## ChatCompletions로 텍스트 생성하기

텍스트를 생성하는 방법은 `ChatCompletionsClient` 클래스를 사용하는 것입니다.
`samples/python/azure_ai_inference/basic.py`의 응답 코드 부분에서, 사용자 역할을 나타내는 부분을 아래처럼 content 매개변수를 변경하세요:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # 선택적 매개변수
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

## 서로 다른 용도의 다양한 프롬프트 유형

이제 프롬프트를 사용해 텍스트를 생성하는 방법을 보았습니다. 다양한 유형의 텍스트를 생성하도록 수정하고 변경할 수 있는 프로그램도 실행 중입니다.

프롬프트는 여러 가지 작업에 사용할 수 있습니다. 예를 들어:

- **텍스트 유형 생성**. 예를 들어, 시를 생성하거나 퀴즈 질문을 생성할 수 있습니다.
- **정보 조회**. 프롬프트를 사용해 '웹 개발에서 CORS가 무엇을 의미하나요?' 같은 정보를 조회할 수 있습니다.
- **코드 생성**. 이메일 검증에 쓰이는 정규 표현식을 개발하거나 웹 앱 같은 전체 프로그램을 생성할 때 프롬프트를 사용할 수 있습니다.

1. 기존 파일을 시작점으로 사용합니다
1. `prompt` 변수를 만들고 예제 코드를 아래와 같이 변경합니다:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

좋습니다, 이제 어떻게 개선할 수 있는지 살펴보겠습니다. 개선하려면 코드를 유연하게 만들어 재료와 레시피 수를 개선하고 변경할 수 있도록 해야 합니다.


1. 코드를 다음과 같이 변경해 봅시다:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````


코드를 시험 실행해 보면 다음과 같이 보일 수 있습니다:
    
```output

3. Remove the saucepan from the heat and stir in the softened gelatin until completely dissolved.
```
