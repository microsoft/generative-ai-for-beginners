# 텍스트 생성 앱 만들기

지금까지 이 커리큘럼을 통해 프롬프트와 "프롬프트 엔지니어링"이라는 분야 등 핵심 개념들을 살펴봤습니다. ChatGPT, Office 365, Microsoft Power Platform 등 다양한 도구들이 프롬프트를 활용해 원하는 작업을 할 수 있도록 지원합니다.

이런 경험을 앱에 추가하려면 프롬프트, 컴플리션 같은 개념을 이해하고 사용할 라이브러리를 선택해야 합니다. 이번 챕터에서 바로 이 부분을 배웁니다.

## 소개

이번 챕터에서는 다음을 배웁니다:

- openai 라이브러리와 핵심 개념을 알아봅니다.
- openai를 활용해 텍스트 생성 앱을 만듭니다.
- 프롬프트, 온도, 토큰 같은 개념을 활용해 텍스트 생성 앱을 만드는 방법을 이해합니다.

## 학습 목표

이 레슨이 끝나면 다음을 할 수 있습니다:

- 텍스트 생성 앱이 무엇인지 설명할 수 있습니다.
- openai를 사용해 텍스트 생성 앱을 만들 수 있습니다.
- 앱에서 토큰 수를 조절하거나 온도를 변경해 다양한 결과를 얻을 수 있습니다.

## 텍스트 생성 앱이란?

일반적으로 앱을 만들면 다음과 같은 인터페이스가 있습니다:

- 명령 기반. 콘솔 앱은 명령어를 입력하면 작업을 수행하는 전형적인 앱입니다. 예를 들어, `git`은 명령 기반 앱입니다.
- 사용자 인터페이스(UI). 일부 앱은 그래픽 사용자 인터페이스(GUI)를 제공해 버튼을 클릭하거나, 텍스트를 입력하거나, 옵션을 선택하는 등 다양한 방식으로 사용할 수 있습니다.

### 콘솔 및 UI 앱의 한계

명령 기반 앱과 비교해보면:

- **제한적입니다.** 아무 명령이나 입력할 수 있는 게 아니라, 앱이 지원하는 명령만 사용할 수 있습니다.
- **언어에 종속적입니다.** 일부 앱은 여러 언어를 지원하지만, 기본적으로 특정 언어에 맞춰져 있습니다. 물론 추가로 언어 지원을 할 수는 있습니다.

### 텍스트 생성 앱의 장점

그렇다면 텍스트 생성 앱은 어떻게 다를까요?

텍스트 생성 앱에서는 더 자유롭게 사용할 수 있습니다. 정해진 명령이나 특정 입력 언어에 제한되지 않고, 자연어로 앱과 상호작용할 수 있습니다. 또 하나의 장점은 이미 방대한 정보로 학습된 데이터 소스와 상호작용한다는 점입니다. 기존 앱은 데이터베이스에 있는 정보만 활용할 수 있지만, 텍스트 생성 앱은 훨씬 더 많은 정보를 다룰 수 있습니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있을까?

만들 수 있는 것은 다양합니다. 예를 들어:

- **챗봇**. 회사나 제품에 대한 질문에 답하는 챗봇을 만들 수 있습니다.
- **도우미**. LLM은 텍스트 요약, 인사이트 추출, 이력서 작성 등 다양한 텍스트 작업에 강점을 보입니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코딩을 도와주는 코드 어시스턴트를 만들 수 있습니다. 예를 들어, GitHub Copilot이나 ChatGPT를 활용해 코드를 작성할 수 있습니다.

## 어떻게 시작할 수 있을까?

LLM과 연동하는 방법은 보통 두 가지입니다:

- API 사용. 프롬프트를 담아 웹 요청을 보내면 생성된 텍스트를 받을 수 있습니다.
- 라이브러리 사용. 라이브러리는 API 호출을 쉽게 할 수 있도록 도와줍니다.

## 라이브러리/SDK

LLM을 다루는 데 널리 쓰이는 라이브러리들이 있습니다:

- **openai**. 이 라이브러리는 모델에 쉽게 연결하고 프롬프트를 보낼 수 있습니다.

그리고 더 높은 수준에서 동작하는 라이브러리들도 있습니다:

- **Langchain**. Langchain은 유명하며 파이썬을 지원합니다.
- **Semantic Kernel**. Semantic Kernel은 마이크로소프트에서 만든 라이브러리로 C#, Python, Java를 지원합니다.

## 다양한 프롬프트 유형, 다양한 용도

이제 프롬프트를 사용해 텍스트를 생성하는 방법을 배웠습니다. 여러분은 이미 실행 중인 프로그램을 수정해서 다양한 종류의 텍스트를 만들어 볼 수도 있습니다.

프롬프트는 여러 가지 작업에 활용할 수 있습니다. 예를 들어:

- **특정 종류의 텍스트 생성**. 예를 들어, 시를 만들거나 퀴즈 문제를 생성할 수 있습니다.
- **정보 검색**. 프롬프트를 사용해 정보를 찾을 수 있습니다. 예를 들어 '웹 개발에서 CORS가 무엇인가요?'와 같은 질문이 가능합니다.
- **코드 생성**. 프롬프트로 코드를 만들 수도 있습니다. 예를 들어 이메일을 검증하는 정규식을 만들거나, 웹 앱과 같은 전체 프로그램을 생성할 수도 있습니다.

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

좋아요, 이제 어떻게 더 개선할 수 있을지 살펴봅시다. 더 나은 결과를 위해서는 코드가 유연하게 동작하도록 만들어야 하며, 재료나 레시피의 개수를 쉽게 변경하고 확장할 수 있어야 합니다.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

코드를 테스트해보면 다음과 같이 될 수 있습니다:

- **쇼핑 리스트 만들기**. 집에 이미 있는 재료를 고려해서 쇼핑 리스트를 만들어보고 싶습니다.

    이 기능을 구현할 때, 하나의 프롬프트로 모두 해결할 수도 있지만, 두 개의 프롬프트로 나누는 방법도 있습니다. 여기서는 두 번째 방법을 시도해보겠습니다. 추가 프롬프트를 넣으려면, 첫 번째 프롬프트의 결과를 두 번째 프롬프트의 컨텍스트로 전달해야 합니다.

    코드에서 첫 번째 프롬프트 결과를 출력하는 부분을 찾아서 아래 코드를 추가하세요:
    
    ```python
    old_prompt_result = response.choices[0].message.content
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
        
    new_prompt = f"{old_prompt_result} {prompt}"
    
    response = client.complete(
        messages=[
            {
                "role": "system",
                "content": "You are a helpful assistant.",
            },
            {
                "role": "user",
                "content": new_prompt,
            },
        ],
        model=model_name,
        # Optional parameters
        temperature=1.,
        max_tokens=1200,
        top_p=1.    
    )
        
    # print response
    print("Shopping list:")
    print(response.choices[0].message.content)
    ```

    참고할 점:

    - 첫 번째 프롬프트의 결과를 새 프롬프트에 추가해서 새로운 프롬프트를 만듭니다:

        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - 새 요청을 보낼 때, 첫 번째 프롬프트에서 사용한 토큰 수를 고려해서 이번에는 `max_tokens`를 1200으로 설정합니다. **토큰 길이에 대한 팁**. 원하는 텍스트를 생성하는 데 필요한 토큰 수를 잘 생각해야 합니다. 토큰은 비용이 들기 때문에, 가능하면 적은 토큰으로 효율적으로 사용하는 것이 좋습니다. 예를 들어, 프롬프트를 더 간결하게 만들어서 토큰을 줄일 수 있을까요?
