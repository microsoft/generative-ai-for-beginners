# 텍스트 생성 앱 만들기

지금까지 이 커리큘럼을 통해 프롬프트와 "프롬프트 엔지니어링"이라는 전체 학문과 같은 핵심 개념들을 보셨습니다. ChatGPT, Office 365, Microsoft Power Platform 등과 같이 상호작용할 수 있는 많은 도구들은 작업을 수행하기 위해 프롬프트를 사용하는 것을 지원합니다.

이러한 경험을 앱에 추가하려면 프롬프트, 완료 등과 같은 개념을 이해하고 함께 작업할 라이브러리를 선택해야 합니다. 바로 이 장에서 배울 내용입니다.

## 소개

이 장에서, 당신은:

- openai 라이브러리와 그 핵심 개념에 대해 배우게 됩니다.
- openai를 사용하여 텍스트 생성 앱을 만듭니다.
- 프롬프트, 온도, 토큰과 같은 개념을 사용하여 텍스트 생성 앱을 만드는 방법을 이해합니다.

## 학습 목표

이 과정을 마치면, 다음을 할 수 있습니다:

- 텍스트 생성 앱이 무엇인지 설명할 수 있습니다.
- openai를 사용하여 텍스트 생성 앱을 만들 수 있습니다.
- 앱을 구성하여 더 많거나 적은 토큰을 사용하고 온도를 변경하여 다양한 출력을 만들 수 있습니다.

## 텍스트 생성 앱이란?

일반적으로 앱을 만들면 다음과 같은 종류의 인터페이스가 있습니다:

- 명령 기반. 콘솔 앱은 명령어를 입력하고 작업을 수행하는 전형적인 앱입니다. 예를 들어 `git`은 명령 기반 앱입니다.
- 사용자 인터페이스(UI). 일부 앱은 버튼 클릭, 텍스트 입력, 옵션 선택 등을 할 수 있는 그래픽 사용자 인터페이스(GUI)를 갖추고 있습니다.

### 콘솔 및 UI 앱의 한계

명령을 입력하는 명령 기반 앱과 비교해보세요:

- <strong>제한적입니다</strong>. 지원하는 명령어만 입력할 수 있고 아무 명령어나 입력할 수 없습니다.
- **언어 특정적입니다**. 일부 앱은 여러 언어를 지원하지만, 기본적으로는 특정 언어에 맞춰 만들어지고 더 많은 언어를 추가할 수 있습니다.

### 텍스트 생성 앱의 장점

그렇다면 텍스트 생성 앱은 어떻게 다를까요?

텍스트 생성 앱에서는 더 많은 유연성이 있습니다. 명령어 집합이나 특정 입력 언어에 제한되지 않고 자연어를 사용해 앱과 상호작용할 수 있습니다. 그리고 이미 방대한 정보 코퍼스로 학습된 데이터 소스와 상호작용하므로 전통적인 앱이 데이터베이스에 제한될 수 있는 것과 달리 가능합니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있나요?

여러 가지를 만들 수 있습니다. 예를 들어:

- <strong>챗봇</strong>. 회사 및 제품에 관한 질문에 답하는 챗봇이 좋은 예가 될 수 있습니다.
- <strong>도우미</strong>. LLM은 텍스트 요약, 텍스트에서 인사이트 얻기, 이력서 등 텍스트 생성 등에 뛰어납니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코딩을 돕는 코드 어시스턴트를 만들 수 있습니다. 예를 들어 GitHub Copilot이나 ChatGPT와 같은 제품을 사용해 코드를 작성할 수 있습니다.

## 어떻게 시작할 수 있을까요?

일반적으로 LLM과 통합하는 방법에는 다음 두 가지 접근 방식이 있습니다:

- API 사용. 여기서는 프롬프트를 포함한 웹 요청을 구성하고 생성된 텍스트를 받습니다.
- 라이브러리 사용. 라이브러리는 API 호출을 캡슐화하여 더 쉽게 사용할 수 있게 해줍니다.

## 라이브러리/SDK

LLM으로 작업할 수 있는 몇 가지 잘 알려진 라이브러리가 있습니다:

- **openai**, 이 라이브러리는 모델에 쉽게 연결하고 프롬프트를 보낼 수 있게 합니다.

그리고 다음과 같이 더 높은 수준에서 작동하는 라이브러리도 있습니다:

- **Langchain**. Langchain은 잘 알려져 있고 Python을 지원합니다.
- **Semantic Kernel**. Semantic Kernel은 Microsoft의 라이브러리로 C#, Python, Java를 지원합니다.

## 다양한 용도의 다양한 프롬프트 유형

이제 프롬프트를 사용해 텍스트를 생성하는 방법을 배웠습니다. 다양한 종류의 텍스트를 생성하도록 수정하고 변경할 수 있는 프로그램도 실행 중입니다.

프롬프트는 여러 가지 작업에 사용할 수 있습니다. 예를 들면:

- **텍스트 유형 생성하기**. 예를 들어, 시, 퀴즈 질문 등 텍스트를 생성할 수 있습니다.
- **정보 조회**. '웹 개발에서 CORS는 무엇을 의미합니까?'와 같은 정보를 찾는 데 프롬프트를 사용할 수 있습니다.
- **코드 생성하기**. 이메일 유효성 검사용 정규식을 작성하거나 웹 앱 같은 전체 프로그램을 생성하는 등의 코드를 생성할 수 있습니다.

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
