## 텍스트 생성 앱이란 무엇인가요?

보통 앱을 만들 때 다음과 같은 인터페이스 유형이 있습니다:

- 명령어 기반. 콘솔 앱은 명령어를 입력하고 작업을 실행하는 전형적인 앱입니다. 예를 들어 `git`는 명령어 기반 앱입니다.
- 사용자 인터페이스(UI). 일부 앱은 버튼 클릭, 텍스트 입력, 옵션 선택 등을 할 수 있는 그래픽 사용자 인터페이스(GUI)를 가지고 있습니다.

### 콘솔 및 UI 앱의 한계

명령어를 입력하는 명령어 기반 앱과 비교해 보겠습니다: 

- <strong>제한적입니다</strong>. 앱이 지원하는 명령어만 입력할 수 있고 모든 명령어를 입력할 수는 없습니다.
- **언어별 구분**. 일부 앱은 여러 언어를 지원하지만, 기본적으로 특정 언어용으로 만들어져 있으며 추가 언어 지원을 할 수는 있습니다.

### 텍스트 생성 앱의 장점

그럼 텍스트 생성 앱은 어떻게 다른가요?

텍스트 생성 앱에서는 더 많은 유연성을 가지며, 정해진 명령어나 특정 입력 언어에 제한되지 않습니다. 대신 자연어를 사용하여 앱과 상호작용할 수 있습니다. 또 다른 장점은 이미 방대한 정보 말뭉치에 대해 학습된 데이터 소스와 상호작용하기 때문에, 전통적인 앱이 데이터베이스에 있는 정보에 제한되는 것과 다릅니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있나요?

많이 만들 수 있습니다. 예를 들어:

- <strong>챗봇</strong>. 회사와 제품에 대해 질문에 답하는 챗봇이 좋은 예가 될 수 있습니다.
- <strong>도우미</strong>. LLM은 텍스트 요약, 텍스트에서 인사이트 추출, 이력서 등 텍스트 생성에 능합니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코드를 작성하는 데 도움을 주는 코드 어시스턴트를 만들 수 있습니다. 예를 들어 GitHub Copilot과 ChatGPT를 이용할 수 있습니다.

- 라이브러리 사용. 라이브러리는 API 호출을 캡슐화하여 더 쉽게 사용할 수 있도록 합니다.

### OpenAI 구성 설정

OpenAI를 사용할 경우 설정 방법은 다음과 같습니다:

```python
client = OpenAI(

  )

```

위에서 다음을 설정합니다:

- `api_key`, OpenAI 대시보드에서 찾은 API 키.
- `deployment`, GPT 버전입니다.

> [!NOTE]

## 텍스트 생성

텍스트를 생성하는 방법은 `responses.create` 메서드를 사용하는 것입니다. 예시는 다음과 같습니다:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

위 코드에서는 응답 객체를 만들고 사용할 모델과 프롬프트를 전달합니다. 그 다음 생성된 텍스트를 출력합니다.

### 채팅 응답

지금까지 단일 프롬프트에서 텍스트를 생성하는 방법을 보았습니다. Responses API는 챗봇에도 적합합니다 - 대화 내역을 쌓기 위해 메시지 목록을 `input`으로 제공합니다. 사용 예시는 다음과 같습니다:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

  > [!NOTE] 이 단계는 Codespaces에서 노트북을 실행하거나 Devcontainer 내에서 실행하는 경우 필요하지 않습니다

## 연습 - 레시피 생성기 만들기

이제 시나리오를 실행했으니, 시연한 시나리오에 맞는 코드를 작성해 봅시다. 이를 위해 다음 단계를 따르세요:

1. 기존 *app.py* 파일을 시작점으로 사용합니다
1. `prompt` 변수를 찾아서 코드를 다음과 같이 변경합니다:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. 다음과 같이 코드를 변경해 봅시다:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

코드를 테스트 실행하는 것은 다음과 같이 보일 수 있습니다:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

- **쇼핑 목록 생성**. 우리는 집에 이미 있는 재료를 고려하여 쇼핑 목록을 생성하고자 합니다.

    이 기능을 위해 하나의 프롬프트에서 모두 해결할 수도 있고 두 개의 프롬프트로 나눌 수도 있습니다. 후자를 시도해 봅시다. 여기에 추가 프롬프트를 제안하지만, 이를 위해서는 이전 프롬프트 결과를 이후 프롬프트의 컨텍스트로 추가해야 합니다.

    첫 번째 프롬프트 결과를 출력하는 부분을 찾아 다음 코드를 아래에 추가하세요:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # 응답 출력
    print("Shopping list:")
    print(response.output_text)
    ```

    다음 사항에 주목하세요:

    - 첫 번째 프롬프트 결과를 새 프롬프트에 추가해 새 프롬프트를 구성합니다: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - 새 요청을 하지만 첫 번째 프롬프트에서 요청한 토큰 수를 고려하기 위해 이번에는 `max_output_tokens`를 1200으로 합니다. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        이 코드를 실행하면 다음과 같은 출력 결과를 얻습니다:

- **토큰 길이에 대한 한마디**. 원하는 텍스트를 생성하는 데 필요한 토큰 수를 고려해야 합니다. 토큰은 비용이 들므로 가능한 한 토큰 수를 절약하는 방향으로 노력해야 합니다. 예를 들어, 프롬프트를 더 간결하게 구성할 수 있는지 생각해보세요.

   토큰 수 변경은 `max_output_tokens` 매개변수를 사용해 조절할 수 있습니다. 예를 들어 100 토큰을 사용하고 싶으면 이렇게 합니다:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **온도 값 실험하기**. 온도는 아직 언급하지 않은 중요한 맥락으로, 프로그램 수행 결과에 영향을 줍니다. 온도 값이 높을수록 출력이 더 무작위적이며, 낮을수록 더 예측 가능합니다. 출력에 변화를 줄지 여부를 고려하세요.

   온도 값을 변경하려면 `temperature` 매개변수를 사용하세요. 예를 들어 온도를 0.5로 설정하려면 이렇게 합니다:

    ```python
    response = client.responses.create(model=deployment, input=messages, temperature=0.5, store=False)
    ```

   > 참고로, 1.0에 가까울수록 출력이 더 다양해집니다.
