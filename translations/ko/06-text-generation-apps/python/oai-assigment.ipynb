## 텍스트 생성 앱이란 무엇인가요?

일반적으로 앱을 만들 때는 다음과 같은 인터페이스가 있습니다:

- 명령 기반. 콘솔 앱은 명령어를 입력하여 작업을 수행하는 전형적인 앱입니다. 예를 들어, `git`은 명령 기반 앱입니다.
- 사용자 인터페이스(UI). 일부 앱은 버튼 클릭, 텍스트 입력, 옵션 선택 등이 가능한 그래픽 사용자 인터페이스(GUI)를 갖습니다.

### 콘솔 및 UI 앱의 한계

명령어를 입력하는 명령 기반 앱과 비교해 보면:

- <strong>제한적입니다</strong>. 앱이 지원하는 명령어만 입력할 수 있으며 아무 명령어나 입력할 수 없습니다.
- **언어 특정입니다**. 일부 앱은 여러 언어를 지원하지만 기본적으로 특정 언어용으로 만들어집니다. 추가 언어 지원이 가능하더라도 그렇습니다.

### 텍스트 생성 앱의 장점

텍스트 생성 앱은 어떻게 다를까요?

텍스트 생성 앱에서는 보다 유연하여 고정된 명령어 집합이나 특정 입력 언어에 제한되지 않습니다. 대신 자연어로 앱과 상호 작용할 수 있습니다. 또 하나의 장점은 대규모 코퍼스에서 학습한 데이터 소스와 상호 작용하기 때문에, 전통적인 앱이 데이터베이스 내에 제한될 수 있는 내용과 달리 더 방대한 정보를 다룰 수 있다는 점입니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있나요?

만들어 볼 수 있는 여러 가지가 있습니다. 예를 들어:

- <strong>챗봇</strong>. 회사와 제품 같은 주제에 대해 질문에 답하는 챗봇을 만들 수 있습니다.
- <strong>헬퍼</strong>. LLM은 텍스트 요약, 인사이트 도출, 이력서 작성 등과 같은 작업에 뛰어납니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코드 작성에 도움을 주는 코드 어시스턴트를 만들 수 있습니다. 예를 들어 GitHub Copilot이나 ChatGPT 같은 제품을 사용할 수 있습니다.

## 어떻게 시작할 수 있나요?

일반적으로 LLM과 통합하는 두 가지 접근 방식이 있습니다:

- API 사용. 여기서는 프롬프트를 포함한 웹 요청을 구성하여 생성된 텍스트를 받아옵니다.
- 라이브러리 사용. 라이브러리는 API 호출을 캡슐화하여 더 쉽게 사용할 수 있게 해줍니다.

### OpenAI 구성 설정

OpenAI를 사용하는 경우 구성 설정 방법은 다음과 같습니다:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']

```

위에서 다음을 설정합니다:

- `api_key`, OpenAI 대시보드에서 찾은 API 키입니다.
- `deployment`, GPT 버전입니다.

> [!NOTE]
> `os.environ`은 환경 변수를 읽는 함수입니다. 예를 들어 `OPENAI_API_KEY` 같은 환경 변수를 읽는 데 사용할 수 있습니다.

## 텍스트 생성

텍스트를 생성하는 방법은 `responses.create` 메서드를 사용하는 것입니다. 예시는 다음과 같습니다:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

위 코드에서는 응답 객체를 만들고 사용할 모델과 프롬프트를 전달합니다. 그런 다음 생성된 텍스트를 출력합니다.

### 채팅 응답

지금까지 단일 프롬프트에서 텍스트를 생성하는 방법을 보았습니다. Responses API는 챗봇에도 적합합니다 - 대화 내역을 구성하는 메시지 리스트를 `input`으로 제공합니다. 사용 예는 다음과 같습니다:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## 연습 - 첫 번째 텍스트 생성 앱 만들기

OpenAI 서비스 설정 및 구성 방법을 배웠으니 이제 첫 번째 텍스트 생성 앱을 만들어봅시다. 앱을 만들려면 다음 단계를 따라 하세요:


1. 가상 환경을 만들고 openai를 설치합니다:

  > [!NOTE] 이 단계는 Codespaces에서 실행하거나 Devcontainer 내에서 실행하는 경우 필요하지 않습니다


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> Windows를 사용 중이라면 `source venv/bin/activate` 대신에 `venv\Scripts\activate`를 입력하세요. 

> [!NOTE]
> OpenAI 키는 https://platform.openai.com/settings/organization/api-keys 에서 `API keys`를 찾아 확인할 수 있습니다. 여기에서 새 키를 생성하고 즉시 값을 복사할 수 있습니다.


1. *app.py* 파일을 만들고 다음 코드를 작성합니다:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    다음과 같은 출력이 표시됩니다:

    ```output
     매우 불행한 _____.

    옛날 옛적에 매우 불행한 인어가 있었습니다.
    ```

## 연습 - 레시피 생성기 만들기

이제 시나리오를 실행해 보았으니, 시연된 시나리오에 맞는 코드를 작성해 봅시다. 이를 위해 다음 단계를 따르세요:

1. 기존의 *app.py* 파일을 시작점으로 사용합니다
1. `prompt` 변수를 찾아서 코드를 다음과 같이 변경하세요:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

1. 다음과 같이 코드를 변경해 봅시다:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

코드를 테스트 실행하는 예는 다음과 같이 생겼을 수 있습니다:
    
```output
No of recipes (for example, 5: 3
List of ingredients (for example, chicken, potatoes, and carrots: milk,strawberries

-Strawberry milk shake: milk, strawberries, sugar, vanilla extract, ice cubes
-Strawberry shortcake: milk, flour, baking powder, sugar, salt, unsalted butter, strawberries, whipped cream        
-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

- **쇼핑 리스트 생성**. 이미 집에 있는 재료를 고려해서 쇼핑 리스트를 생성하고자 합니다.

    이 기능은 하나의 프롬프트에서 모두 해결하거나 두 개로 나눌 수 있습니다. 후자 방식을 시도해 보겠습니다. 여기서는 추가 프롬프트를 제안하는데, 이를 위해 첫 번째 프롬프트 결과를 두 번째 프롬프트의 컨텍스트로 넣어야 합니다.

    첫 번째 프롬프트 결과를 출력하는 코드 부분을 찾아 아래 코드를 추가합니다:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # 응답 출력
    print("Shopping list:")
    print(response.output_text)
    ```

    다음 사항을 참고하세요:

    - 첫 번째 프롬프트 결과를 새로운 프롬프트에 추가하여 새로운 프롬프트를 구성합니다:
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - 새 요청을 하지만 첫 번째 요청의 토큰 수를 고려하여 이번에는 `max_output_tokens`를 1200으로 지정합니다.

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
        ```  

        위 코드를 실행하면 다음과 같은 출력이 나옵니다:

- **토큰 길이에 대하여**. 원하는 텍스트를 생성하는데 필요한 토큰 수를 고려해야 합니다. 토큰은 비용이 들기 때문에 가능한 토큰 수를 절약하는 것이 좋습니다. 예를 들어, 프롬프트 문장을 간략하게 해서 토큰 수를 줄일 수 있는지 생각해 보세요.

   토큰 수는 `max_output_tokens` 매개변수를 사용해 조정할 수 있습니다. 예를 들어 100 토큰을 사용하려면 다음과 같이 지정합니다:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **온도 실험하기**. 온도는 아직 언급하지 않았지만 프로그램 성능에 중요한 역할을 합니다. 온도가 높을수록 출력이 더 무작위적이고, 낮을수록 더 예측 가능합니다. 출력에 변화를 원할지 고려하세요.

   온도를 조절하려면 `temperature` 매개변수를 사용합니다. 예를 들어 0.5로 설정하려면:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```

   > 참고로 1.0에 가까울수록 출력이 더 다양해집니다.
