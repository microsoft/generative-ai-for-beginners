## 텍스트 생성 앱이란?

일반적으로 앱을 만들 때는 다음과 같은 인터페이스가 있습니다:

- 명령 기반. 콘솔 앱은 명령을 입력하고 작업을 수행하는 전형적인 앱입니다. 예를 들어 `git`은 명령 기반 앱입니다.
- 사용자 인터페이스 (UI). 일부 앱은 버튼 클릭, 텍스트 입력, 옵션 선택 등을 할 수 있는 그래픽 사용자 인터페이스(GUI)를 갖추고 있습니다.

### 콘솔 및 UI 앱은 제한적입니다

명령을 입력하는 명령 기반 앱과 비교해보면:

- <strong>제한적입니다</strong>. 앱이 지원하는 명령만 입력할 수 있으며 아무 명령이나 입력할 수 없습니다.
- **언어에 특화되어 있습니다**. 일부 앱은 여러 언어를 지원하지만 기본적으로 특정 언어를 위해 만들어졌고, 더 많은 언어 지원을 추가할 수도 있습니다.

### 텍스트 생성 앱의 장점

텍스트 생성 앱은 어떻게 다를까요?

텍스트 생성 앱에서는 더 유연하며, 정해진 명령 집합이나 특정 입력 언어에 제한되지 않습니다. 대신 자연어를 사용해 앱과 상호작용할 수 있습니다. 또한 이미 방대한 코퍼스로 훈련된 데이터 소스와 상호작용하기 때문에 기존 앱이 데이터베이스 내 정보에 제한될 수 있는 것과 달리 장점이 있습니다.

### 텍스트 생성 앱으로 무엇을 만들 수 있을까요?

다양한 것을 만들 수 있습니다. 예를 들어:

- <strong>챗봇</strong>. 회사 및 제품 관련 질문에 답하는 챗봇이 좋은 예입니다.
- <strong>도우미</strong>. LLM은 텍스트 요약, 텍스트에서 통찰 얻기, 이력서 같은 텍스트 작성 등에 탁월합니다.
- **코드 어시스턴트**. 사용하는 언어 모델에 따라 코드를 작성하는 데 도움을 주는 코드 어시스턴트를 만들 수 있습니다. 예를 들어 GitHub Copilot이나 ChatGPT 같은 제품을 사용할 수 있습니다.

### Azure 구성 설정

Azure OpenAI를 사용하는 경우 구성 설정 방법은 다음과 같습니다. Responses API는 Azure OpenAI <strong>v1 엔드포인트</strong>에서 제공되므로 `OpenAI` 클라이언트를 `<your-endpoint>/openai/v1/`로 지정합니다:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

위 코드에서는 다음을 설정합니다:

- `api_key`: Azure Portal에서 찾은 API 키입니다.
- `base_url`: `/openai/v1/`가 추가된 Azure OpenAI 엔드포인트입니다. API 키 옆에 Azure Portal에서 찾을 수 있습니다. v1 엔드포인트를 사용하면 더 이상 `api_version`을 전달할 필요가 없습니다.
- `deployment`: Foundry 포털에서 생성한 모델 배포 이름입니다.

> [!NOTE]
> `os.environ`은 환경변수를 읽는 매핑입니다. 이를 사용해 `AZURE_OPENAI_API_KEY`와 `AZURE_OPENAI_ENDPOINT` 같은 환경 변수를 읽을 수 있습니다.

## 텍스트 생성하기

텍스트를 생성하려면 `responses.create` 메서드를 사용합니다. 예시는 다음과 같습니다:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

위 코드에서는 응답 객체를 만들고 사용하려는 모델과 프롬프트를 전달합니다. 그런 다음 생성된 텍스트를 출력합니다.

### 채팅 응답

Responses API는 단일 턴 텍스트 생성과 다중 턴 챗봇 모두에 적합합니다 - 대화 구축을 위해 `input` 리스트에 메시지를 더 많이 제공하기만 하면 됩니다:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

## 실습 - 첫 번째 텍스트 생성 앱

이제 Azure OpenAI 서비스를 설정하고 구성하는 방법을 배웠으니 첫 번째 텍스트 생성 앱을 만들어 봅시다. 앱을 만들려면 다음 단계를 따르세요:


1. 가상 환경을 만들고 openai를 설치합니다:

  > [!NOTE] 이 단계는 Codespaces에서 실행하거나 Devcontainer 내에서 실행하는 경우 필요하지 않습니다


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> Windows를 사용하는 경우 `source venv/bin/activate` 대신 `venv\Scripts\activate`를 입력하세요. 

> [!NOTE]
> https://portal.azure.com/ 에서 `Open AI`를 검색하고 `Open AI resource`를 선택한 다음 `Keys and Endpoint`를 선택하여 `Key 1` 값을 복사하여 Azure OpenAI 키를 찾으세요.


1. *app.py* 파일을 만들고 다음 코드를 작성합니다:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````

    ```output

    ```

## 연습 - 레시피 생성기 만들기

이제 시나리오를 진행해 보았으니, 시연된 시나리오에 맞는 코드를 작성해 봅시다. 그렇게 하려면 다음 단계를 따르세요:

1. 기존 *app.py* 파일을 시작점으로 사용하세요
1. `prompt` 변수를 찾아 그 코드를 다음과 같이 변경하세요:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````
