## Що таке додаток для генерації тексту?

Зазвичай, коли ви створюєте додаток, він має якийсь інтерфейс, наприклад:

- Командний рядок. Консольні додатки — це типові додатки, де ви вводите команду, і вони виконують завдання. Наприклад, `git` — це командний додаток.

### Консольні додатки та UI мають обмеження

Порівняйте це з командним додатком, де ви вводите команду:

- **Він обмежений**. Ви не можете просто ввести будь-яку команду, лише ті, які підтримує додаток.

### Переваги додатків для генерації тексту

Чим же відрізняється додаток для генерації тексту?

У додатку для генерації тексту у вас більше гнучкості: ви не обмежені набором команд або конкретною мовою вводу. Натомість ви можете використовувати природну мову для взаємодії з додатком. Ще одна перевага в тому, що ви працюєте з джерелом даних, яке навчалось на величезному корпусі інформації, тоді як традиційний додаток може бути обмеженим даними лише в базі.

## Генерація тексту

Для генерації тексту використовується метод `responses.create`. Ось приклад:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

У наведеному коді ми створюємо об’єкт відповіді, передаємо модель, яку хочемо використати, і підказку. Потім виводимо згенерований текст.

### Відповіді у чаті

Поки що ви бачили, як генерувати текст із однієї підказки. Responses API також добре підходить для чатботів — ви передаєте список повідомлень у параметр `input` для формування розмови. Ось приклад використання:

## Вправа - створення генератора рецептів

Тепер, коли ми розіграли сценарій, давайте напишемо код, що відповідає продемонстрованому сценарію. Для цього виконайте такі кроки:

1. Використайте існуючий файл *app.py* як відправну точку
1. Знайдіть змінну `prompt` і змініть її код на наступний:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Покращення шляхом додавання фільтра та списку покупок
