## Вправa - створити генератор рецептів

Тепер, коли ми відтворили сценарій, напишемо код, що відповідає продемонстрованому сценарію. Для цього виконайте такі кроки:

1. Використайте наявний файл *app.py* як відправну точку
1. Знайдіть змінну `prompt` та замініть її код на наступний:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Чудово, давайте подивимось, як ми можемо покращити ситуацію. Щоб покращити ситуацію, ми хочемо переконатися, що код є гнучким, тому інгредієнти та кількість рецептів можна покращувати і змінювати.

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Покращення шляхом додавання фільтра і списку покупок

Тепер у нас є робочий додаток, здатний генерувати рецепти, і він гнучкий, оскільки залежить від введених користувачем даних — як кількості рецептів, так і використаних інгредієнтів.

Для подальшого покращення ми хочемо додати наступне:

- **Створити список покупок**. Ми хочемо створити список покупок, враховуючи те, що вже є вдома.

- **Експерименти з параметром temperature**. Температура — це те, про що ми раніше не говорили, але що сильно впливає на роботу програми. Чим вища температура, тим випадковіший буде результат.
