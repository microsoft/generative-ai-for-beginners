## 什麼是文字生成應用程式？

通常當你建立應用程式時，它會有某種介面，如下所示：

- 命令式(Command-based)。主控台(Console)應用程式是典型的命令式應用程式，你鍵入命令後，它會執行任務。例如，`git` 是一個命令式應用程式。
- 使用者介面(UI)。有些應用程式擁有圖形使用者介面(GUI)，用戶能點擊按鈕、輸入文本、選擇選項等。

### 主控台與 UI 應用程式的限制

比較一個命令式應用程式，你輸入指令：

- <strong>有其限制</strong>。你不能隨意輸入任何命令，只能輸入應用程式所支援的命令。
- <strong>語言特定</strong>。有些應用程式支援多種語言，但預設為特定語言，雖然你可以額外加入語言支援。

### 文字生成應用程式的好處

那麼，文字生成應用程式有何不同？

在文字生成應用程式中，你有更大的彈性，不受限於一組命令或特定語言輸入。你可以使用自然語言來與應用程式互動。另一個好處是你正與一個基於龐大資料集訓練的資料來源互動，而傳統應用程式可能只能依賴資料庫中的內容。

### 用文字生成應用程式可以建立什麼？

你可以建立許多應用。例如：

- <strong>聊天機器人</strong>。像是回答公司和產品相關問答的聊天機器人就很適合。
- <strong>輔助工具</strong>。大型語言模型擅長摘要文本、從文本中獲取見解、產生文本如履歷等。
- <strong>程式碼助理</strong>。依據你所使用的語言模型，你可以建立幫助撰寫程式碼的助理。例如，你可以使用 GitHub Copilot 或 ChatGPT 幫助寫程式。

## 生成文字

生成文字的方法是使用 `responses.create`。範例如下：

```python

print(response.output_text)
```

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````
