## 什麼是文字生成應用程式？

通常你建立的應用程式會有某種介面，如以下這些：

- 指令式。命令列應用程式是典型的程式，你輸入指令它會執行任務。例如，`git` 是一個指令式應用程式。
- 使用者介面（UI）。有些應用程式有圖形使用者介面（GUI），你可以點擊按鈕、輸入文字、選擇選項等。

### 指令列和 UI 應用程式的限制

跟一個你輸入指令的指令式應用程式相比：

- <strong>它是有侷限的</strong>。你不能隨意輸入任何指令，只能輸入程式支援的指令。
- <strong>語言專屬性</strong>。有些應用程式支援多種語言，但預設是針對特定語言開發，即使你可以新增更多語言支援。

### 文字生成應用程式的好處

那文字生成應用程式有什麼不同呢？

文字生成應用程式給你更多的彈性，你不受限於固定指令或特定輸入語言。你可以使用自然語言與應用程式互動。另一個好處是，因為你已經在與訓練過大量語料庫的資料來源互動，而傳統應用程式可能只侷限於資料庫中的內容。

### 我可以用文字生成應用程式做什麼？

你可以建立許多不同的應用程式，例如：

- <strong>聊天機器人</strong>。像是回答關於你的公司和產品的問題的聊天機器人會很適合。
- <strong>助理工具</strong>。大型語言模型很擅長摘要文字、從文字中獲得洞察、產生像履歷表這樣的文字等。
- <strong>程式碼助理</strong>。根據你使用的語言模型，你可以建立一個協助你撰寫程式碼的程式碼助理。例如，你可以使用 GitHub Copilot 或 ChatGPT 來協助編寫程式碼。

## 產生文字

產生文字的方法是使用 `responses.create` 方法。下面是一個範例：

```python
prompt = "Complete the following: Once upon a time there was a"

print(response.output_text)
```

上面程式碼中，我們建立一個 response 物件，並傳入想使用的模型和提示詞。然後我們列印生成的文字。

## 練習—你的第一個文字生成應用程式

現在我們學會了如何設定及組態 OpenAI 服務，是時候建立你的第一個文字生成應用程式。要建立你的應用程式，請按照以下步驟進行：


1. 建立虛擬環境並安裝 openai：

  > [!NOTE] 如果您在 Codespaces 或 Devcontainer 中執行此筆記本，則不需要此步驟


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> 如果你使用的是 Windows，請使用 `venv\Scripts\activate`，而不是 `source venv/bin/activate`。

> [!NOTE]
> 前往 https://platform.openai.com/settings/organization/api-keys 查找你的 OpenAI 金鑰，並搜尋 `API keys`。你可以在那裡建立新金鑰並立即複製其值。


1. 建立一個 *app.py* 檔案並加入以下程式碼：


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    你應該會看到如下的輸出：

    ```output
     非常不快樂的 _____。

    從前有一位非常不快樂的人魚。
    ```
