പ്രകൃതി ഭാഷാ പ്രോസസ്സിംഗ് പദ്ധതികൾക്കായി പ്രോംപ്റ്റുകൾ രൂപകൽപ്പന ചെയ്‌തും മെച്ചപ്പെടുത്തലും നടത്തുകയായിരുന്ന പ്രക്രിയയാണ് പ്രോംപ്റ്റ് എഞ്ചിനീയറിംഗ്. ഇത് ശരിയായ പ്രോംപ്റ്റുകൾ തിരഞ്ഞെടുക്കൽ, അവയുടെ പാരാമീറ്ററുകൾ ക്രമപ്പെടുത്തൽ, അവയുടെ പ്രകടനം വിലയിരുത്തൽ എന്നിവയെയാണ് ഉൾക്കൊള്ളുന്നതെന്ന്. NLP മോഡലുകളിൽ ഉയർന്ന കൃത്യതയും കാര്യക്ഷമതയും കൈവരിക്കാനായി പ്രോംപ്റ്റ് എഞ്ചിനീയറിംഗ് അത്യന്താപേക്ഷിതമാണ്. ഈ വിഭാഗത്തിൽ, OpenAI മോഡലുകൾ ഉപയോഗിച്ച് പ്രോംപ്റ്റ് എഞ്ചിനീയറിംഗിന്റെ അടിസ്ഥാനങ്ങൾ പരിശോധിക്കും.


### അഭ്യാസം 1: ടോക്കനൈസേഷൻ  
OpenAIയിൽ നിന്നുള്ള ഒരു ഓപ്പൺ-സോഴ്‌സ് ഫാസ്റ്റ് ടോക്കനൈസറായ tiktoken ഉപയോഗിച്ച് ടോക്കനൈസേഷൻ അന്വേഷിക്കുക  
കൂടുതൽ ഉദാഹരണങ്ങൾക്ക് [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) കാണുക.  


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### വ്യായാമം 2: OpenAI API കീ സജ്ജീകരണം സാധുവോ എന്ന് പരിശോധിക്കുക

നിങ്ങളുടെ OpenAI എന്റ്പോയിന്റ് ശരിയായി സജ്ജീകരിച്ചിട്ടുണ്ടോ എന്ന് സ്ഥിരീകരിക്കാൻ താഴെയുള്ള കോഡ് പ്രവർത്തിപ്പിക്കുക. കോഡ് ഒരു ലളിതമായ അടിസ്ഥാന പ്രോമ്പ്റ്റ് പരീക്ഷിക്കുകയും പൂർത്തീകരണം സാധുവാണോ എന്ന് പരിശോധിക്കുകയും ചെയ്യുന്നു. ഇൻപുട്ട് `oh say can you see` എന്നത് `by the dawn's early light..` എന്ന തരത്തിലായിരിക്കും പൂർത്തീകരിക്കുന്നത്.


````python
# Uses the OpenAI client with the Responses API.
# See https://platform.openai.com/docs/api-reference/responses

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### അഭ്യാസം 3: കൃതി സൃഷ്ടികള്‍
ഏതെങ്കിലും നിലവിലുള്ളതല്ലാത്ത വിഷയം സംബന്ധിച്ചോ, അല്ലെങ്കില്‍ അതിന്റെ മുമ്പ് പരിശീലന ഡാറ്റാസെറ്റിനകത്ത് ഇല്ലാത്ത (പുതിയതായി) വിഷയംകൊണ്ട് ഒരു പ്രോമ്പ്റ്റിന് പരിഹാരങ്ങള്‍ നല്‍കണമെന്ന് LLM-നെ ചോദിക്കുകയാണെങ്കില്‍ സംഭവിക്കുന്നതെന്ത് എന്ന് അന്വേഷിക്കുക. വ്യത്യസ്തമായ ഒരു പ്രോമ്പ്റ്റ് അല്ലെങ്കില്‍ വ്യത്യസ്ത മോഡല്‍ ഉപയോഗിച്ചാല്‍ പ്രതികരണം എങ്ങിനെയാണെന്നാണ് നോക്കുക.


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### വ്യായാമം 4: നിർദ്ദേശം അടിസ്ഥാനമാക്കിയുള്ളത്

ആ പ്രാഥമിക ഉള്ളടക്കവുമായി ബന്ധപ്പെട്ട ഒരു നിർദ്ദേശം നൽകാൻ "prompt" വ്യത്യസ്തം ഉപയോഗിക്കുക.

ഇവിടെ നമ്മൾ മോഡലോട് രണ്ടാം ക്ലാസ് വിദ്യാർത്ഥിക്ക് വേണ്ടി വാചകത്തെ സംക്ഷിപ്തമാക്കണമെന്ന് ചോദിക്കുന്നു


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### അഭ്യാസം 5: സങ്കീർണ്ണ പ്രോംപ്റ്റ്  
സിസ്റ്റം, ഉപയോക്താവ്, അസിസ്റ്റന്റ് സന്ദേശങ്ങൾ ഉള്ള ഒരു അഭ്യർത്ഥന ശ്രമിക്കുക  
സിസ്റ്റം അസിസ്റ്റൻറിന്റെ പാരിസ്ഥിതികം സജ്ജീകരിക്കുന്നു  
ഉപയോക്താവ് & അസിസ്റ്റന്റ് സന്ദേശങ്ങൾ ബഹുരൂപ സംഭാഷണ പാരിസ്ഥിതികം നൽകുന്നു  

സിസ്റ്റം പാരിസ്ഥിതികത്തിൽ അസിസ്റ്റന്റിന്റെ വ്യക്തിത്വം "വ്യഞ്ജനപരമായ" എന്നാണ് സജ്ജീകരിച്ചിരിക്കുന്നത് എന്നത് ശ്രദ്ധിക്കുക.  
മറ്റൊരു വ്യക്തിത്വ പാരിസ്ഥിതികം ഉപയോഗിച്ച് ശ്രമിക്കൂ. അല്ലെങ്കിൽ ഇൻപുട്ട്/ഔട്ട്പുട്ട് സന്ദേശങ്ങളുടെ വ്യത്യസ്ത ശ്രേണി പരീക്ഷിക്കൂ  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
