# പ്രോപ്റ്റ് എഞ്ചിനീയറിംഗിലേക്കുള്ള പരിചയം
സ്വാഭാവിക ഭാഷാ പ്രോസസ്സിംഗ് പ്രവർത്തനങ്ങൾക്ക് പ്രോപ്റ്റുകൾ രൂപകൽപ്പന ചെയ്ത് മെച്ചപ്പെടുത്തുന്നതാണ് പ്രോപ്റ്റ് എഞ്ചിനീയറിംഗ്. ശരിയായ പ്രോപ്റ്റുകൾ തിരഞ്ഞെടുക്കുക, അവയുടെ പാരാമീറ്ററുകൾ ക്രമീകരിക്കുക, അവയുടെ പ്രകടനം വിലയിരുത്തുക എന്നിവ ഇതിൽ ഉൾപ്പെടുന്നു. NLP മോഡലുകളിൽ ഉയർന്ന കൃത്യതയിലും കാര്യക്ഷമതയിലും എത്താൻ പ്രോപ്റ്റ് എഞ്ചിനീയറിംഗ് നിർണായകമാണ്. ഈ വിഭാഗത്തിൽ, ഓപ്പൺഎഐ മോഡലുകൾ ഉപയോഗിച്ച് പ്രോപ്റ്റ് എഞ്ചിനീയറിംഗിന്റെ അടിസ്ഥാനങ്ങൾ ഞങ്ങൾ പഠിക്കും.


### അഭ്യാസം 1: ടോക്കൺവത്കരണം
ഓപ്പൺഎഐയുടെ ഒരു ഓപ്പൺ-സോഴ്‌സ് ഫാസ്റ്റ് ടോക്കൺവത്കരിക്കാരായ tiktoken ഉപയോഗിച്ച് ടോക്കൺവത്കരണം അന്വോഷിക്കുക
കൂടുതൽ ഉദാഹരണംകൾക്കായി [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) കാണുക.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### വ്യായാം 2: OpenAI API കീ ക്രമീകരണം പരിശോധിക്കുക

നിങ്ങളുടെ OpenAI എൻഡ്‌പോയിന്റ് ശരിയായി ക്രമീകരിച്ചിട്ടുണ്ടോ എന്ന് പരിശോധിക്കാൻ താഴെയുള്ള കോഡ് പ്രവർത്തിപ്പിക്കുക. കോഡ് ഒരു ലളിതമായ അടിസ്ഥാന പ്രോംപ്റ്റ് പരീക്ഷിച്ച് പൂരണമാകുന്നത് ശരിയാണോ എന്ന് സ്ഥിരീകരിക്കുന്നു. `oh say can you see` എന്ന ഇൻപുട്ട് `by the dawn's early light..` എന്ന രീതിയിൽ പൂരണമാകണം.


````python
# Uses the OpenAI client with the Responses API.
# See https://platform.openai.com/docs/api-reference/responses

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### വ്യായാമം 3: ഫാബ്രിക്കേഷൻസ്
ഒന്നുള്ളതല്ലാത്ത വിഷയത്തെക്കുറിച്ചോ അതിന്റെ മുൻപു പരിശീലിപ്പിച്ച ഡാറ്റാസെറ്റിന് പുറത്തുള്ള (ഇತ್ತೀಚത്തെ) വിഷയങ്ങളെക്കുറിച്ചോ പ്രേരണ ചോദിക്കുമ്പോൾ എൽഎൽഎം എന്താണ് ചെയ്യുന്നത് എന്ന് പരിശോധിക്കുക. വ്യത്യസ്തമായ ഒരു പ്രേരണയോ വ്യത്യസ്ത മോഡലോ ഉപയോഗിച്ചാൽ പ്രതികരണം എങ്ങനെ മാറുന്നു എന്ന് കാണുക.


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### അഭ്യാസം 4: നിർദ്ദേശം അടിസ്ഥാനമാക്കി 
പ്രാഥമിക ഉള്ളടക്കം സജ്ജമാക്കാൻ "text" മാരകാ ചൊവ്വയുപയോഗിക്കുക 
ആ പ്രാഥമിക ഉള്ളടക്കത്തെ ബന്ധപ്പെട്ട ഒരു നിർദ്ദേശം നൽകാൻ "prompt" മാരകാ ചൊവ്വയുപയോഗിക്കുക.

ഇവിടെ ഞങ്ങൾ മോഡലിനെ ടെക്സ്റ്റ് രണ്ടാം ക്ലാസ്സ് വിദ്യാർത്ഥിക്ക് സംക്ഷിപ്തമാക്കാൻ ആവശ്യപ്പെടുന്നു


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### അഭ്യാസം 5: സങ്കീർണ്ണ പ്രോംപ്റ്റ്  
സിസ്റ്റം, ഉപയോക്താവ്, അസിസ്റ്റന്റ് സന്ദേശങ്ങൾ ഉൾക്കൊള്ളുന്ന ഒരു അഭ്യർത്ഥന പരീക്ഷിക്കുക  
സിസ്റ്റം അസിസ്റ്റന്റെ संदർഭം നിജപ്പെടുത്തുന്നു  
ഉപയോക്താവും അസിസ്റ്റന്റും സന്ദേശങ്ങൾ ബഹുസഞ്ചാരി സംഭാഷണ संदർഭം നൽകുന്നു  

സിസ്റ്റം संदർഭത്തിൽ അസിസ്റ്റന്റ് വ്യക്തിത്വം " സാമ്രാജ്ഞികമായത്" എന്ന് എങ്ങനെ സജ്ജീകരിച്ചിരിക്കുകയാണ് ശ്രദ്ധിക്കുക.  
വ്യത്യസ്ത വ്യക്തിത്വ संदർഭം പരീക്ഷിക്കുക. അല്ലെങ്കിൽ വ്യത്യസ്തമായ ഒരു ഇൻപുട്ട്/ഔട്ട്പുട്ട് സന്ദേശങ്ങളുടെ പരമ്പര പരീക്ഷിക്കുക  


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````

### വ്യായാമം: നിങ്ങളുടെ അകത്രം ശ്രദ്ധിക്കുക
മുകളിൽ നൽകിയ ഉദാഹരണങ്ങൾ പുതിയ പ്രോമ്പ്റ്റുകൾ സൃഷ്ടിക്കാൻ ഉപയോഗിക്കാവുന്ന മാതൃകകൾ നൽകുന്നു (സാധാരണ, സങ്കീർണ്ണം, നിർദ്ദേശം എന്നിവ) - നമ്മൾ ചർച്ച ചെയ്‌ത മറ്റ് ആശയങ്ങൾ പോലുള്ള ഉദാഹരണങ്ങൾ, സൂചനകൾ എന്നിവ പരിശോധിക്കാൻ മറ്റ് വ്യായാമങ്ങൾ സൃഷ്ടിക്കാൻ ശ്രമിക്കുക.
