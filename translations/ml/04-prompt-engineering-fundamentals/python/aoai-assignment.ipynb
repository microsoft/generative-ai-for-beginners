# പ്രാംപ്‍ട്ട് എഞ്ചിനീയറിങ്ങിലേക്കുള്ള പരിചയം
സ്വാഭാവിക ഭാഷാ പ്രോസസ്സിംഗ് ടാസ്കുകൾക്കായുള്ള പ്രാംപ്്റ്റുകൾ രൂപകൽപ്പന ചെയ്ത് അവ എങ്ങനെ മെച്ചപ്പെടുത്താമെന്ന് പരിശോധിക്കുന്ന പ്രക്രിയയാണ് പ്രാംപ്്റ്റ് എഞ്ചിനീയറിങ്. ശരിയായ പ്രാംപ്്റ്റുകൾ തിരഞ്ഞെടുക്കുന്നത്, അവയുടെ പരാമീറ്ററുകൾ ക്രമീകരിക്കുന്നതും പരീക്ഷിക്കുന്നതും ഇതിൽ ഉൾപ്പെടുന്നു. NLP മോഡലുകളിൽ ഉയർന്ന വൈദഗ്ധ്യവും കാര്യക്ഷമതയും ഉറപ്പാക്കുന്നതിനായി പ്രാംപ്്റ്റ് എഞ്ചിനീയറിങ് അത്യന്താപേക്ഷിതമാണ്. ഈ വിഭാഗത്തിൽ, OpenAI മോഡലുകൾ ഉപയോഗിച്ച് പ്രാംപ്്റ്റ് എഞ്ചിനീയറിങ്ങിന്റെ അടിസ്ഥാനങ്ങൾ പരിശോധിക്കാം.


### വ്യായാമം 1: ടോക്കനൈസേഷൻ
OpenAI യുടെ ഒരു ഓപ്പൺ-സോഴ്‌സ് വേഗത്തിൽ പ്രവർത്തിക്കുന്ന ടോക്കനൈസറായ tiktoken ഉപയോഗിച്ച് ടോക്കനൈസേഷൻ പരീക്ഷിക്കുക
കൂടുതൽ ഉദാഹരണങ്ങൾക്കായി [OpenAI Cookbook](https://github.com/openai/openai-cookbook/blob/main/examples/How_to_count_tokens_with_tiktoken.ipynb?WT.mc_id=academic-105485-koreyst) കാണുക.


````python
# EXERCISE:
# 1. Run the exercise as is first
# 2. Change the text to any prompt input you want to use & re-run to see tokens

import tiktoken

# Define the prompt you want tokenized
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

# Set the model you want encoding for
encoding = tiktoken.encoding_for_model("gpt-4o")

# Encode the text - gives you the tokens in integer form
tokens = encoding.encode(text)
print(tokens);

# Decode the integers to see what the text versions look like
[encoding.decode_single_token_bytes(token) for token in tokens]
````

### അഭ്യാസം 2: OpenAI API കീയുടെ സജ്ജീകരണം സ്ഥിരീകരിക്കുക

താഴെ കൊടുത്ത കോഡ് പ്രവർത്തിപ്പിച്ച് നിങ്ങളുടെ OpenAI എന്റ്പോയിന്റ് ശരിയായി സജ്ജീകരിച്ചിട്ടുണ്ടോ എന്നത് പരിശോധിക്കുക. ഈ കോഡ് ഒരു ലളിതമായ അടിസ്ഥാന പ്രോംപ്റ്റ് പരിശോധിച്ച് പൂർണ്ണത സ്ഥിരീകരിക്കും. `oh say can you see` എന്ന ഇൻപുട്ട് `by the dawn's early light..` എന്ന രീതിയിൽ പൂർത്തീകരിക്കണം.


````python
# Uses the OpenAI client against the Azure OpenAI (Microsoft Foundry) v1 endpoint
# with the Responses API. See https://aka.ms/openai/start

import os
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']

def get_completion(prompt):
    response = client.responses.create(
        model=deployment,
        input=prompt,
        temperature=0, # this is the degree of randomness of the model's output
        max_output_tokens=1024,
        store=False,
    )
    return response.output_text

## ---------- Call the helper method

### 1. Set primary content or prompt text
text = f"""
oh say can you see
"""

### 2. Use that in the prompt template below
prompt = f"""
```{text}```
"""

## 3. Run the prompt
response = get_completion(prompt)
print(response)
````

### വ്യായാമം 3: നിര്‍മ്മിതികള്‍
പ്രീ-ട്രെയിനഡ് ഡാറ്റാസെറ്റിന്റെ (കൂടുതല്‍ சமീപകാലത്തെ) പുറത്ത് ഉള്ളതിനാല്‍ ല്‍എല്‍എം അറിവിലില്ലാത്ത ഒരു വിഷയത്തെക്കുറിച്ചോ അല്ലാത്ത ഒരു വിഷയം കണക്കിലെടുത്ത് പ്രാപ്തി ചോദിച്ചാല്‍ എന്ത് സംഭവിക്കുന്നുവെന്ന് അന്വേഷിക്കുക. വ്യത്യസ്തമായ ഒരു പ്രാപ്തി, അല്ലെങ്കില്‍ വ്യത്യസ്ത മോഡല്‍ ഉപയോഗിച്ചാല്‍ പ്രതികരണം എങ്ങനെ മാറുന്നു എന്ന് കാണുക.


````python

## Set the text for simple prompt or primary content
## Prompt shows a template format with text in it - add cues, commands etc if needed
## Run the completion 
text = f"""
generate a lesson plan on the Martian War of 2076.
"""

prompt = f"""
```{text}```
"""

response = get_completion(prompt)
print(response)
````

### വ്യായാമം 4: നിർദേശ അടിസ്ഥാനമെౖതം 
പ്രാഥമിക ഉള്ളടക്കമുള്ള "text" വേരിയബിൾ ഉപയോഗിക്കുക 
പ്രാഥമിക ഉള്ളടക്കത്തോട് ബന്ധപ്പെട്ട ഒരു നിർദേശം നൽകാൻ "prompt" വേരിയബിൾ ഉപയോഗിക്കുക.

ഇവിടെ ഞങ്ങൾ മോഡലിനോട് ടെക്സ്റ്റ് രണ്ടാം ക്ലാസ് വിദ്യാർത്ഥിക്ക് സംഗ്രഹിച്ചു നൽകാൻ ചോദിക്കുന്നു


````python
# Test Example
# https://platform.openai.com/playground/p/default-summarize

## Example text
text = f"""
Jupiter is the fifth planet from the Sun and the \
largest in the Solar System. It is a gas giant with \
a mass one-thousandth that of the Sun, but two-and-a-half \
times that of all the other planets in the Solar System combined. \
Jupiter is one of the brightest objects visible to the naked eye \
in the night sky, and has been known to ancient civilizations since \
before recorded history. It is named after the Roman god Jupiter.[19] \
When viewed from Earth, Jupiter can be bright enough for its reflected \
light to cast visible shadows,[20] and is on average the third-brightest \
natural object in the night sky after the Moon and Venus.
"""

## Set the prompt
prompt = f"""
Summarize content you are provided with for a second-grade student.
```{text}```
"""

## Run the prompt
response = get_completion(prompt)
print(response)
````

### വ്യായാമം 5: സങ്കീർണ്ണമായ പ്രോംപ്റ്റ് 
സിസ്റ്റം, ഉപയോക്താവ്, അസിസ്റ്റന്റ് സന്ദേശങ്ങൾ ഉൾക്കൊള്ളുന്ന ഒരു അഭ്യർത്ഥന ശ്രമിക്കുക 
സിസ്റ്റം അസിസ്റ്റന്റിന്റെ കോൺടെക്സ്റ്റ് ക്രമീകരിക്കുന്നു
ഉപയോക്താവിന്റെയും അസിസ്റ്റന്റിന്റെയും സന്ദേശങ്ങൾ ബഹുമടിക്കുള്ള സംഭാഷണത്തിന്റെ കോൺടെക്സ്റ്റ് നൽകുന്നു

സിസ്റ്റം കോൺടെക്സ്റ്റിൽ അസിസ്റ്റന്റിന്റെ വ്യക്തിത്വം "വ്യംഗ്യപരമായ" എന്ന് ക്രമീകരിച്ചിട്ടുള്ളത് ശ്രദ്ധിക്കുക. 
മറ്റൊരു വ്യക്തിത്വ കോൺടെക്സ്റ്റ് ഉപയോഗിക്കാൻ ശ്രമിക്കുക. അല്ലെങ്കിൽ വ്യത്യസ്തമായ ഇൻപുട്ട്/ഔട്ട്പുട്ട് സന്ദേശങ്ങളുടെ ഒരു പരമ്പര ശ്രമിക്കുക


````python
response = client.responses.create(
    model=deployment,
    input=[
        {"role": "system", "content": "You are a sarcastic assistant."},
        {"role": "user", "content": "Who won the world series in 2020?"},
        {"role": "assistant", "content": "Who do you think won? The Los Angeles Dodgers of course."},
        {"role": "user", "content": "Where was it played?"}
    ],
    store=False,
)
print(response.output_text)
````
