## മെറ്റാ ഫാമിലി മോഡലുകൾ 

ഈ പാഠത്തിൽ, മെറ്റാ ഫാമിലി അല്ലെങ്കിൽ "ലാമ ഹേർഡ്" എന്നതിൽ നിന്നുള്ള 2 മോഡലുകൾ - ലാമ 3.1, ലാമ 3.2 എന്നിവ പരിശോധിക്കും 

ഈ മോഡലുകൾ വ്യത്യസ്ത വേർഷനുകളിൽ ലഭ്യവും [Microsoft Foundry Models catalog](https://ai.azure.com/catalog/models?WT.mc_id=academic-105485-koreyst)ൽ ലഭ്യവുമാണ്.

> **ശ്രദ്ധിക്കുക:** GitHub Models ജൂലൈ 2026 അന്ത്യത്തിൽ വിരമിക്കുന്നു. AI മോഡലുകളുമായി പ്രോട്ടോടൈപ്പ് ചെയ്യാൻ [Microsoft Foundry Models](https://learn.microsoft.com/azure/ai-foundry/model-inference/overview?WT.mc_id=academic-105485-koreyst) ഉപയോഗിക്കുന്നതിനു കൂടുതൽ വിവരങ്ങൾ ഇവിടെ ഉണ്ടായിരിക്കുന്നു.

മോഡൽ വേർഷനുകൾ: 
- ലാമ 3.1 - 70B ഇൻസ്ട്രക്ട് 
- ലാമ 3.1 - 405B ഇൻസ്ട്രക്ട് 
- ലാമ 3.2 - 11B വിശൻ ഇൻസ്ട്രക്ട് 
- ലാമ 3.2 - 90B വിശൻ ഇൻസ്ട്രക്ട് 

*ശ്രദ്ധിക്കുക: ലാമ 3 മൈക്രോസോഫ്റ്റ് ഫൗണ്ട്രി മോഡലുകളിലും ലഭ്യമാണ്, പക്ഷെ ഈ പാഠത്തിൽ അത് ഉൾപ്പെടുത്തപ്പെടില്ല*



## ല്ലാമ 3.1 

405 ബില്യൺ പാരാമീറ്ററുകളോടെ, ല്ലാമ 3.1 ഓപ്പൺ സോഴ്സ് LLM വിഭാഗത്തിലേക്ക് പെടുന്നു. 

ഈ മോഡ് മുൻകാല റിലീസ് ല്ലാമ 3-നേക്കാൾ അപ്ഗ്രേഡ് നൽകിയതാണ്: 

- വലുത어진 കോൺടക്സ്റ്റ് വിൻഡോ - 8k ടോക്കൺസിന് പകരം 128k ടോക്കൺസ് 
- വലുത어진 പരമാവധി ഔട്ട്‌പുട്ട് ടോക്കൺസ് - 2048-ന്റെ പകരം 4096 
- മെച്ചപ്പെട്ട ബഹുഭാഷാ പിന്തുണ - പരിശീലന ടോക്കൺസിന്റെ വർദ്ധനവിന് കാരണമാകുന്നു 

ജിയൻഎഐ ആപ്ലിക്കേഷനുകൾ നിർമ്മിക്കുമ്പോൾ ല്ലാമ 3.1 കൂടുതൽ സങ്കീർണ്ണ ഉപയോഗ കേസുകൾ കൈകാര്യം ചെയ്യാൻ സാധിക്കുന്നതാണ്: 
- नेटീവ് ഫങ്ഷൻ കോളിംഗ് - LLM വർക്‌ഫ്ലോവിനപ്പുറം പുറത്തുള്ള ഉപകരണങ്ങളും ഫങ്ഷനുകളും വിളിക്കാനുള്ള കഴിവ്
- മെച്ചപ്പെട്ട RAG പ്രകടനം - വലുത어진 കോൺടക്സ്റ്റ് വിൻഡോവിനെ തുടർന്ന് 
- സിന്തറ്റിക് ഡാറ്റ ജനറേഷൻ - ഫൈൻ-ട്യൂണിംഗ് പോലുള്ള പ്രവർത്തനങ്ങൾക്ക് ഫലപ്രദമായ ഡാറ്റ സൃഷ്ടിക്കാനുള്ള കഴിവ് 



### നേറ്റീവ് ഫംഗ്ഷൻ കോളിങ് 

ല്ലാമ 3.1 ഫംഗ്ഷൻ അല്ലെങ്കിൽ ടൂൾ കോളുകൾ കൂടുതൽ ഫലപ്രദമായി നടത്താനായി സൂക്ഷ്മമായ പരിശീലനം ലഭിച്ചിട്ടുണ്ട്. ഉപയോക്താവിന്റെയും പ്രോംപ്റ്റിനും അടിസ്ഥാനത്തിൽ ഉപയോഗിക്കേണ്ടതുണ്ടെന്ന് മോഡൽ തിരിച്ചറിയാൻ കഴിയുന്ന രണ്ട് നിർമാണ ടൂളുകളും ഇതിന് ഉണ്ട്. ആ ടൂളുകൾ: 

- **ബ്രേവ് സേർച്ച്** - വെബ് സേർച്ച് നടത്തുമ്പോൾ കാലാവസ്ഥ പോലുള്ള പുതുക്കപ്പെട്ട വിവരങ്ങൾ നേടാൻ ഉപയോഗിക്കാവുന്ന ഉദ്‌യോகம் 
- **വോൾഫ്രാം ആൽഫ** - കൂടുതൽ സങ്കീർണ്ണ ഗണിത കണക്കുകൂട്ടലുകൾക്കായി ഉപയോഗിക്കാവുന്ന ടൂൾ, അതുകൊണ്ട് സ്വയം ഫംഗ്ഷനുകൾ എഴുതേണ്ടതില്ല. 

നിങ്ങൾക്ക് താങ്കളുടെ സ്വന്തം ഇഷ്ടാനുസൃത ടൂളുകളും സൃഷ്‌ടിക്കാം, LLM അവയെ വിളിച്ചുപയോഗിക്കാം. 

താഴെ നൽകിയ കോഡ് ഉദാഹരണത്തിൽ: 

- ലഭ്യമായ ടൂളുകൾ (brave_search, wolfram_alpha) സിസ്റ്റം പ്രോംപ്റ്റിൽ നിർവചിക്കുന്നു. 
- ഒരു ഉപയോഗகரന്റെ പ്രോംപ്റ്റ് സെന്റ് ചെയ്യുന്നു, ഒരു പ്രത്യേക നഗരത്തിലെ കാലാവസ്ഥയെക്കുറിച്ച് ചോദിക്കുന്നു. 
- LLM Brave Search ടൂളിലേക്ക് ടൂൾ കോളായി പ്രതികരിക്കും, ഇങ്ങനെ കാണപ്പെടും `<|python_tag|>brave_search.call(query="Stockholm weather")` 

*കുറിപ്പ്: ഉദാഹരണം ടൂൾ കോളിംഗ് മാത്രം ചെയ്യുന്നു, ഫലങ്ങൾ ലഭിക്കാൻ നിങ്ങൾക്ക് Brave API പേജിൽ ഒരു വേഗ പടിയിലുളള അക്കൗണ്ടും ഫംഗ്ഷന്റെയും നിർവചനവും വേണം* 


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import AssistantMessage, SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Meta-Llama-3.1-405B-Instruct"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)


tool_prompt=f"""
<|begin_of_text|><|start_header_id|>system<|end_header_id|>

Environment: ipython
Tools: brave_search, wolfram_alpha
Cutting Knowledge Date: December 2023
Today Date: 23 July 2024

You are a helpful assistant<|eot_id|>
"""

messages = [
    SystemMessage(content=tool_prompt),
    UserMessage(content="What is the weather in Stockholm?"),

]

response = client.complete(messages=messages, model=model_name)

print(response.choices[0].message.content)




```

### ല്ലാമ 3.2 

ഒരു LLM ആണെങ്കിലും, ല്ലാമ 3.1യുടെ ഒരു പരിധി മൾട്ടിമൊഡാലിറ്റി ആണ്. അഥവാ, ചിത്രങ്ങൾ പോലെയുള്ള വ്യത്യസ്ത തരത്തിലുള്ള ഇൻപുട്ടുകൾ പ്രോംപ്റ്റായി ഉപയോഗിച്ച് അവയ്ക്ക് പ്രതികരിക്കാനുള്ള കഴിവ്. ഈ കഴിവാണ് ല്ലാമ 3.2യുടെ പ്രധാന കഴിവുകളിൽ ഒന്നും. ഈ കഴിവുകൾ ഉൾപ്പെടുന്നു: 

- മൾട്ടിമൊഡാലിറ്റി - ടെക്സ്റ്റും ഇമേജും പ്രോംപ്റ്റുകളിൽ വിലയിരുത്താൻ കഴിവുള്ളത് 
- ചെറുതിൽ നിന്നു മധ്യനില വരെയുള്ള വേരിയേഷൻകൾ (11B ਅਤੇ 90B) - ഇത് ഫ്ലെക്‌സിബിള്‍ ഡിപ്ലോയ്മെന്റ് ഓപ്ഷനുകൾ നൽകുന്നു, 
- ടെക്സ്റ്റ് മാത്രം വരുന്ന വേരിയേഷൻകൾ (1B, 3B) - ഈ മോഡൽ എഡ്ജ്/മൊബൈൽ ഉപകരണങ്ങളിൽ ഡിപ്ലോയ് ചെയ്യാനാകും, കൂടാതെ കുറഞ്ഞ ലാറ്റൻസി നൽകുന്നു 

മൾട്ടിമോഡൽ പിന്തുണ ഓപ്പൺ സോഴ്സ് മോഡലുകളുടെ ലോകത്ത് വലിയൊരു മുന്നേറ്റമാണ്. താഴെ കൊടുത്ത കോഡ് ഉദാഹരണം ല്ലാമ 3.2 90B-നിന്ന് ചിത്രത്തിന്റെ വിശകലനം നേടാൻ ഒരു ചിത്രം കൂടാതെ ടെക്സ്റ്റ് പ്രോംപ്റ്റും എടുക്കുന്നു. 

### ല്ലാമ 3.2-വുമായുള്ള മൾട്ടിമൊഡൽ പിന്തുണ


```python
%pip install azure-core
%pip install azure-ai-inference
```

```python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import (
    SystemMessage,
    UserMessage,
    TextContentItem,
    ImageContentItem,
    ImageUrl,
    ImageDetailLevel,
)
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]
model_name = "Llama-3.2-90B-Vision-Instruct"


client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

response = client.complete(
    messages=[
        SystemMessage(
            content="You are a helpful assistant that describes images in details."
        ),
        UserMessage(
            content=[
                TextContentItem(text="What's in this image?"),
                ImageContentItem(
                    image_url=ImageUrl.load(
                        image_file="sample.jpg",
                        image_format="jpg",
                        detail=ImageDetailLevel.LOW)
                ),
            ],
        ),
    ],
    model=model_name,
)

print(response.choices[0].message.content)
```
