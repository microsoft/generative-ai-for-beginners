# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾ കെട്ടിപ്പടുക്കുക

നിങ്ങൾ ഇതുവരെ ഈ പാഠ്യപദ്ധതിയിൽ കണ്ടിട്ടുള്ളത് പ്രാധാന്യപെടുത്തേണ്ട പ്രൊംപ്റ്റുകൾ പോലുള്ള അടിസ്ഥാന ആശയങ്ങളാണുള്ളത്, മിക്കപ്പോഴും "പ്രൊംപ്‌റ്റ് എഞ്ചിനിയറിങ്" എന്ന മുഴുവൻ ശാസ്ത്രശാഖയും ഉണ്ട്. ChatGPT, Office 365, Microsoft Power Platform തുടങ്ങിയ പല ഉപകരണങ്ങളും, നിങ്ങൾ അബ്ജു കഴിയുന്ന പ്രൊംപ്റ്റുകൾ ഉപയോഗിച്ച് എന്തെങ്കിലും സാധ്യമാക്കാൻ സഹായിക്കുന്നു.

ഇത്തരമൊരു അനുഭവം ഒരു ആപ്പിൽ ചേർക്കാൻ, പ്രൊംപ്റ്റുകളും, പൂർത്തീകരണങ്ങളും അറിയിച്ചു എന്ന് മനസ്സിലാക്കേണ്ടതാണ്, കൂടാതെ പ്രവർത്തിക്കാൻ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കേണ്ടതാണ്. അതാണ് നിങ്ങൾ ഈ അധ്യായത്തിൽ പഠിക്കാനിരിക്കുന്നത്.

## പരിചയം

ഈ അധ്യായത്തിൽ, നിങ്ങൾ:

- openai ലൈബ്രറിയെയും അതിന്റെ അടിസ്ഥാന ആശയങ്ങളെയും കുറിച്ച് അറിയുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- പ്രൊംപ്റ്റ്, താപനില, ടോക്കണുകൾ പോലുള്ള ആശയങ്ങൾ ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ എങ്ങനെ ഉപയോഗിക്കാമെന്ന് മനസ്സിലാക്കുക.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠത്തിന്റെ അവസാനം, നിങ്ങൾക്ക് കഴിയും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- നിങ്ങളുടെ ആപ്പിൽ കൂടുതലോ കുറവോ ടോക്കണുകൾ ഉപയോഗിക്കാൻ കൂടാതെ താപനിലയും മാറ്റാൻ ക്രമീകരിക്കുക, ഇതനുസരിച്ചു വ്യത്യസ്ത ഔട്ട്പുട്ട് ലഭിക്കും.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി നിങ്ങൾ ഒരു ആപ്പ് നിർമ്മിക്കുമ്പോൾ താഴെപ്പറയുന്ന പോലൊരു ഇന്റർഫേസ് ഉണ്ടാകും:

- കമാൻഡ് അധിഷ്ഠിതം. കോൺസോൾ ആപ്പുകൾ ജോലി നിർവഹിക്കാൻ കമാൻഡ് ടൈപ്പ് ചെയ്യുന്നതാണ് സാമാന്യ, ഉദാഹരണത്തിന് `git` ഒരു കമാൻഡ് അടിസ്ഥാന ആപ്പാണ്.
- യൂസർ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾക്ക് ഗ്രാഫിക്കൽ യൂസർ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ടാകും, അവിടെ നിങ്ങൾ ബട്ടണുകൾ ക്ലിക്കുചെയ്യാൻ, ടെക്സ്റ്റ് ഇന്റർ ചെയ്താലോ, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കലോ ചെയ്യാം.

### കോൺസോൾ മെന്നും UI ആപ്പുകളും പരിമിതമാണെന്ന്

കമാൻഡ് ടൈപ്പ് ചെയ്യുന്ന കമാൻഡ് അധിഷ്ഠിത ആപ്പുമായി താരതമ്യം ചെയ്യുക:

- **അപരിമിതം**. നിങ്ങൾക്ക് ഏതൊരു കമാൻഡും ടൈപ്പ് ചെയ്തത് സാധ്യമല്ല, ആപ്പ് പിന്തുണയ്ക്കുന്ന കമാൻഡുകൾ മാത്രമേ സാധ്യമാകൂ.
- **ഭാഷാ പ്രത്യേകം**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കാം, പക്ഷെ സ്വാഭാവികമായി ആപ്പ് ഒരു പ്രത്യേക ഭാഷയ്ക്ക് മാത്രമാണ് നിർമ്മിച്ചിരിക്കുന്നത്, അതിലേക്ക് കൂടുതൽ ഭാഷയും ചേർക്കാം.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളുടെ ഗുണങ്ങൾ

എങ്ങനെ ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് വ്യത്യസ്തമാണ്?

ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ നിങ്ങൾക്ക് കൂടുതൽ സൗകര്യമുണ്ട്, നിങ്ങൾ കമാൻഡുകൾക്ക് അല്ലെങ്കിൽ ഒരു പ്രത്യേക ഇൻപുട്ട് ഭാഷയിലേക്കു മാത്രമേ പരിമിതരായിരിക്കൂ. പകരം, സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ആശയവിനിമയം നടത്താം. മറ്റൊരു ഗുണം, നിങ്ങൾ ഇതിനകം വലിയ വിവരശേഖരത്തിൽ പരിശീലിച്ച ഡാറ്റാ ഉള്ളടക്കവുമായി ഇടപഴകിയതാണ്, പരമ്പരാഗത ആപ്പുകൾക്ക് ഡാറ്റാബേസിൽ ഉള്ളതിൽ പരിമിതമായിത്തീരും.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ നിങ്ങൾ എന്ത് നിർമ്മിക്കാം?

പലതും നിർമ്മിക്കാം, ഉദാഹരണത്തിന്:

- **ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനി ഉദാ. ഉൽപ്പന്നങ്ങളെക്കുറിച്ചു ചോദ്യങ്ങൾക്ക җаваൽ നൽകുന്ന ചാറ്റ്ബോട്ട് നല്ലൊരു തിരഞ്ഞെടുപ്പാകും.
- **സഹായി**. LLMകൾ ടെക്സ്റ്റ് സംഗ്രഹിക്കൽ, ടെക്സ്റ്റിൽ നിന്നുള്ള ഉൾക്കാഴ്ചകൾ നേടൽ, റിസ്യൂമെ പോലുള്ള ടെക്സ്റ്റ് ഉത്പാദനം എന്നിവയിൽ മികച്ചതാണ്.
- **കോഡ് അസിസ്റ്റന്റ്**. നിങ്ങൾ ഉപയോഗിക്കുന്ന ഭാഷ മോഡലിന്റെ അടിസ്ഥാനത്തിൽ, കോഡ് എഴുതാൻ സഹായിക്കുന്ന കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉല്പന്നവും ChatGPT ഉം കോഡ് എഴുതാൻ സഹായിക്കുന്നു.

## എങ്ങനെ ആരംഭിക്കാം?

LLM യുമായുള്ള ഇന്റഗ്രേഷൻ സാധ്യമാക്കാനുള്ള രണ്ട് പ്രധാന സമീപനങ്ങളുണ്ട്:

- API ഉപയോഗിക്കുക. ഇവിടെ നിങ്ങൾ പ്രൊംപ്റ്റുമായി വെബ് അഭ്യർത്ഥനകൾ നിർമ്മിക്കുകയും, ജനറേറ്റുചെയ്ത ടെക്സ്റ്റ് തിരിച്ചു ലഭിക്കുകയും ചെയ്യുന്നു.
- ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API കോൾകൾ ഉൾപ്പെടുത്തി ഉപയോഗിക്കാൻ എളുപ്പമാക്കുന്നു.

## ലൈബ്രറികൾ/SDKകൾ

LLMs ഉപയോഗിക്കാൻ ചില പ്രശസ്ത ലൈബ്രറികളുണ്ട്:

- **openai**, ഈ ലൈബ്രറി നിങ്ങളുടെ മോഡലുമായി ബന്ധപ്പെടാനും പ്രൊംപ്റ്റുകൾ അയയ്ക്കാനും എളുപ്പമാണ്.

പിന്നെ ഉയർന്ന തലത്തിലുള്ള ചില ലൈബ്രറികളും ഉണ്ട്:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python പിന്തുണയുള്ളത്.
- **Semantic Kernel**. Semantic Kernel മൈക്രോസോഫ്റ്റിന്റെ ലൈബ്രറി ആണ്, C#, Python, Java എന്നീ ഭാഷകൾ പിന്തുണയ്ക്കുന്നു.

## openai ഉപയോഗിച്ച് ആദ്യ ആപ്പ്

ആദ്യ ആപ്പ് എങ്ങനെ നിർമ്മിക്കാമെന്ന് നോക്കാം, എത്ര ലൈബ്രറികൾ ആവശ്യമാണ്, മറ്റു വിവരങ്ങളും.

### openai ഇൻസ്റ്റാൾ ചെയ്യുക

  > [!NOTE] ഇത് Codespaces-ൽ അല്ലെങ്കിൽ Devcontainer-ൽ പ്രവർത്തിപ്പിക്കുമ്പോൾ ആവശ്യമായില്ല


OpenAI അല്ലെങ്കിൽ Azure OpenAI-യുമായി സംവദിക്കാൻ നിരവധി ലൈബ്രറികൾ ഉണ്ട്. C#, Python, JavaScript, Java പോലുള്ള പല പ്രോഗ്രാമിങ് ഭാഷകളും ഉപയോഗിക്കാം.  
നാം `openai` Python ലൈബ്രറി ഉപയോഗിക്കാൻ തിരഞ്ഞെടുത്തു, അതിനാൽ `pip` ഉപയോഗിച്ച് ഇൻസ്റ്റാൾ ചെയ്യും.

```bash
pip install openai
```

Codespaces അല്ലെങ്കിൽ Dev Container-ൽ ഈ നോട്ട്ബുക്ക് പ്രവർത്തിപ്പിക്കുന്നില്ലെങ്കിൽ, [Python](https://www.python.org/) നിങ്ങളുടെ യന്ത്രത്തിൽ ഇൻസ്റ്റാൾ ചെയ്യേണ്ടതുണ്ട്.

### ഒരു റിസോഴ്സ് സൃഷ്ടിച്ച് API കീ കണ്ടെത്തുക

നിങ്ങൾ ഇതുവരെ ചെയ്തിട്ടില്ലെങ്കിൽ, താഴെപ്പറയുന്ന നടപടിക്രമങ്ങൾ നടത്തണം:

- OpenAI-ൽ അക്കൗണ്ട് സൃഷ്ടിക്കുക <https://platform.openai.com/signup>.
- ഇനി, നിങ്ങളുടെ API കീകൾ നേടുക <https://platform.openai.com/api-keys>. 

>[!NOTE]
> നിങ്ങളുടെ API കീ കോഡിൽ നിന്ന് വേർതിരിക്കാൻ വിലമതിക്കുന്നു. ഇത് ഏർത്ത് എടുക്കാൻ എന്വയോൺമെന്റ് വേരിയബിളുകൾ ഉപയോഗിക്കാം.
> - `.env` ഫയിൽൽ `OPENAI_KEY` എന്ന ympäristömuuttujaksi asetetaan API avain. ഇത് മുൻപുള്ള വർക്ക്‌സിന്റെ ഏർപ്പെടുത്തലുകൾ പൂർത്തിയാക്കിയാൽ ഒരുങ്ങിച്ചേർത്ത് വരും.
> - API കീ ഒരു മാത്രം ലഭ്യമാകും. അതുകൊണ്ട് ശരിയായി കോപ്പി ചെയ്തോയെന്ന് സ്ഥിരീകരിക്കൽ അത്യാവശ്യം. പ്രവർത്തിക്കാതെ പക്ഷേ, കീ നീക്കി മറ്റൊന്ന് സൃഷ്ടിക്കുക.


### OpenAI ക്രമീകരണം സെറ്റ് ചെയ്യുക

OpenAI ഉപയോഗിക്കുകയാണെങ്കിൽ, ക്രമീകരണമെങ്ങനെ നടത്തുമെന്ന്:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-4o-mini"
```

മുകളിൽ നമ്മൾ ക്രമീകരിക്കുന്നത്:

- `api_key`, ഇത് OpenAI ഡാഷ്ബോർഡിൽ കണ്ടെത്തുന്ന നിങ്ങൾയുടെ API കീ ആണ്.
- `deployment`, ഇത് GPT പതിപ്പാണ്.

> [!NOTE]
> `os.environ` ഒരു ഫംഗ്ഷൻ ആണ്, ഇത് എന്വയോൺമെന്റ് മാറ്റികൾ വായിക്കാൻ ഉപയോഗിക്കുന്നു. നിങ്ങൾക്ക് ഇതുപയോഗിച്ച് `OPENAI_API_KEY` പോലുള്ള മാറ്റികൾ വായിക്കാം.

> ശ്രദ്ധിക്കൂ, നിങ്ങളുടെ LLM അനിശ്ചിതസ്വഭാവമുള്ളതാണ്, അതിനാൽ ഓരോ പ്രാവശ്യം പ്രോഗ്രാം റൺ ചെയ്യുമ്പോഴും വ്യത്യസ്ത ഫലങ്ങൾ ലഭിക്കാം.

ഉത്തമം, നമുക്ക് കാര്യങ്ങൾ മെച്ചപ്പെടുത്തുന്നത് എങ്ങനെ എടുക്കാമെന്ന് നോക്കാം. കാര്യങ്ങൾ മെച്ചപ്പെടുത്താൻ, കോഡ് ഫ്ലെക്സിബിളായിരിക്കണം, അതായത് ഘടകങ്ങളും വിഭവങ്ങളുടെ എണ്ണം മാറ്റാനും മെച്ചപ്പെടുത്താനും കഴിയണം.


1. താഴെ കൊടുത്ത രീതിയിൽ കോഡ് മാറ്റാം:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

```output

-Strawberry milk: milk, strawberries, sugar, vanilla extract
```

    ഈ ഫംഗ്ഷണിലിറ്റിക്ക്, ഒറ്റ പ്രോംപ്റ്റിൽ എല്ലാം പരിഹരിക്കാൻ ശ്രമിക്കാമോ, അല്ലെങ്കിൽ രണ്ട് പ്രോംപ്റ്റുകളായി വിഭജിക്കാമോ. രണ്ടാം മാർഗം പരീക്ഷിക്കാം. ഇവിടെ ഒരു അധിക പ്രോംപ്റ്റ് ചേർക്കാൻ നിർദ്ദേശിക്കുന്നു, പക്ഷേ അത് പ്രവർത്തിക്കാൻ, മുൻ പ്രോംപ്റ്റിന്റെ ഫലം രണ്ടാം പ്രോംപ്റ്റിന്റെ കണ്ടെക്സ്റ്റായി നൽകണം.
