# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾ നിർമ്മിക്കുക

ഈ പാഠ്യപദ്ധതിയിൽ ഇതുവരെ നിങ്ങൾ കണ്ടത് പോലെ പ്രോംപ്റ്റുകൾ പോലുള്ള അടിസ്ഥാന ആശയങ്ങളും "പ്രോംപ്റ്റ് എഞ്ചിനീയറിംഗ്" എന്ന മുഴുവൻ ശാഖയും ഉണ്ട്. ChatGPT, Office 365, Microsoft Power Platform എന്നിവ പോലുള്ള പല ഉപകരണങ്ങളും നിങ്ങൾ പ്രോംപ്റ്റുകൾ ഉപയോഗിച്ച് എന്തെങ്കിലും നേടാൻ സഹായിക്കുന്നു.

അത്തരത്തിലുള്ള അനുഭവം ഒരു ആപ്പിൽ ചേർക്കാൻ, പ്രോംപ്റ്റുകൾ, പൂർത്തീകരണങ്ങൾ എന്നിവ പോലുള്ള ആശയങ്ങൾ മനസ്സിലാക്കുകയും പ്രവർത്തിക്കാൻ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കുകയും വേണം. ഇതാണ് ഈ അധ്യായത്തിൽ നിങ്ങൾ പഠിക്കാനിരിക്കുന്നതും.

## പരിചയം

ഈ അധ്യായത്തിൽ, നിങ്ങൾ:

- openai ലൈബ്രറിയും അതിന്റെ അടിസ്ഥാന ആശയങ്ങളും പഠിക്കും.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കും.
- പ്രോംപ്റ്റ്, താപനില, ടോക്കണുകൾ പോലുള്ള ആശയങ്ങൾ ഉപയോഗിച്ച് ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എങ്ങനെ നിർമ്മിക്കാമെന്ന് മനസ്സിലാക്കും.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠത്തിന്റെ അവസാനം, നിങ്ങൾക്ക് കഴിയും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- നിങ്ങളുടെ ആപ്പ് കൂടുതൽ അല്ലെങ്കിൽ കുറവ് ടോക്കണുകൾ ഉപയോഗിച്ച്, കൂടാതെ താപനില മാറ്റി വ്യത്യസ്ത ഔട്ട്പുട്ട് ലഭ്യമാക്കാൻ ക്രമീകരിക്കുക.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി ഒരു ആപ്പ് നിർമ്മിക്കുമ്പോൾ അതിന് താഴെപ്പറയുന്ന പോലുള്ള ഇന്റർഫേസ് ഉണ്ടാകും:

- കമാൻഡ് അടിസ്ഥാനമാക്കിയുള്ളത്. കൺസോൾ ആപ്പുകൾ സാധാരണയായി നിങ്ങൾ കമാൻഡ് ടൈപ്പ് ചെയ്ത് ഒരു ജോലി നിർവഹിക്കുന്ന ആപ്പുകളാണ്. ഉദാഹരണത്തിന്, `git` ഒരു കമാൻഡ് അടിസ്ഥാനമാക്കിയുള്ള ആപ്പാണ്.
- ഉപയോക്തൃ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾക്ക് ഗ്രാഫിക്കൽ ഉപയോക്തൃ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ട്, അവിടെ നിങ്ങൾ ബട്ടണുകൾ ക്ലിക്ക് ചെയ്യുകയും, ടെക്സ്റ്റ് നൽകുകയും, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കുകയും ചെയ്യുന്നു.

### കൺസോൾ, UI ആപ്പുകൾക്ക് പരിധിയുണ്ട്

നിങ്ങൾ കമാൻഡ് ടൈപ്പ് ചെയ്യുന്ന കമാൻഡ് അടിസ്ഥാനമാക്കിയുള്ള ആപ്പുമായി താരതമ്യം ചെയ്യുക:

- **പരിധിയുള്ളതാണ്**. നിങ്ങൾക്ക് ഏതെങ്കിലും കമാൻഡ് ടൈപ്പ് ചെയ്യാൻ കഴിയില്ല, ആപ്പ് പിന്തുണയ്ക്കുന്നവ മാത്രമേ ടൈപ്പ് ചെയ്യാനാകൂ.
- **ഭാഷാ പ്രത്യേകമാണ്**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കുന്നു, പക്ഷേ സാധാരണയായി ആപ്പ് ഒരു പ്രത്യേക ഭാഷക്കായി നിർമ്മിച്ചിരിക്കുന്നു, കൂടുതൽ ഭാഷാ പിന്തുണ ചേർക്കാൻ കഴിയുന്നുണ്ടെങ്കിലും.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളുടെ ഗുണങ്ങൾ

അപ്പോൾ ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എങ്ങനെ വ്യത്യസ്തമാണ്?

ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ നിങ്ങൾക്ക് കൂടുതൽ സ്വാതന്ത്ര്യമുണ്ട്, കമാൻഡുകളുടെ ഒരു സെറ്റിലോ ഒരു പ്രത്യേക ഇൻപുട്ട് ഭാഷയിലോ നിങ്ങൾ പരിമിതരാകുന്നില്ല. പകരം, നിങ്ങൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ഇടപഴകാം. മറ്റൊരു ഗുണം, നിങ്ങൾ ഇതിനകം വലിയ വിവരശേഖരത്തിൽ പരിശീലനം നേടിയ ഡാറ്റാ സ്രോതസ്സുമായി ഇടപഴകുകയാണ്, പരമ്പരാഗത ആപ്പുകൾക്ക് ഡാറ്റാബേസിൽ ഉള്ളതിൽ പരിമിതമായിരിക്കാം.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ ഞാൻ എന്ത് നിർമ്മിക്കാം?

നിങ്ങൾക്ക് നിരവധി കാര്യങ്ങൾ നിർമ്മിക്കാം. ഉദാഹരണത്തിന്:

- **ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനി, അതിന്റെ ഉൽപ്പന്നങ്ങൾ എന്നിവയെക്കുറിച്ച് ചോദ്യങ്ങൾക്ക് മറുപടി നൽകുന്ന ഒരു ചാറ്റ്ബോട്ട് നല്ലൊരു തിരഞ്ഞെടുപ്പാകും.
- **സഹായി**. LLMകൾ ടെക്സ്റ്റ് സംഗ്രഹിക്കൽ, ടെക്സ്റ്റിൽ നിന്നുള്ള洞察ങ്ങൾ നേടൽ, റിസ്യൂമുകൾ പോലുള്ള ടെക്സ്റ്റ് നിർമ്മാണം എന്നിവയിൽ മികച്ചവയാണ്.
- **കോഡ് അസിസ്റ്റന്റ്**. നിങ്ങൾ ഉപയോഗിക്കുന്ന ഭാഷാ മോഡലിന്റെ അടിസ്ഥാനത്തിൽ, നിങ്ങൾക്ക് കോഡ് എഴുതാൻ സഹായിക്കുന്ന ഒരു കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉൽപ്പന്നവും ChatGPTയും കോഡ് എഴുതാൻ സഹായിക്കാൻ ഉപയോഗിക്കാം.

## എങ്ങനെ തുടങ്ങാം?

നിങ്ങൾക്ക് സാധാരണയായി താഴെപ്പറയുന്ന രണ്ട് സമീപനങ്ങൾ ഉപയോഗിച്ച് LLM-നുമായി സംയോജിപ്പിക്കേണ്ടതുണ്ട്:

- API ഉപയോഗിക്കുക. ഇവിടെ നിങ്ങൾ നിങ്ങളുടെ പ്രോംപ്റ്റ് ഉപയോഗിച്ച് വെബ് അഭ്യർത്ഥനകൾ നിർമ്മിച്ച് ജനറേറ്റഡ് ടെക്സ്റ്റ് തിരികെ ലഭിക്കും.
- ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API കോൾസ് ഉൾക്കൊള്ളിച്ച് അവ ഉപയോഗിക്കാൻ എളുപ്പമാക്കുന്നു.

## ലൈബ്രറികൾ/SDKകൾ

LLM-കളുമായി പ്രവർത്തിക്കാൻ ചില പ്രശസ്തമായ ലൈബ്രറികൾ ഉണ്ട്:

- **openai**, ഈ ലൈബ്രറി നിങ്ങളുടെ മോഡലുമായി ബന്ധപ്പെടാനും പ്രോംപ്റ്റുകൾ അയയ്ക്കാനും എളുപ്പമാക്കുന്നു.

പിന്നീട് ഉയർന്ന തലത്തിൽ പ്രവർത്തിക്കുന്ന ലൈബ്രറികളും ഉണ്ട്:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python പിന്തുണയ്ക്കുന്നു.
- **Semantic Kernel**. Semantic Kernel മൈക്രോസോഫ്റ്റിന്റെ ഒരു ലൈബ്രറിയാണ്, C#, Python, Java ഭാഷകൾ പിന്തുണയ്ക്കുന്നു.

## openai ഉപയോഗിച്ച് ആദ്യ ആപ്പ്

നമുക്ക് നമ്മുടെ ആദ്യ ആപ്പ് എങ്ങനെ നിർമ്മിക്കാമെന്ന്, എത്ര ലൈബ്രറികൾ വേണമെന്ന്, മറ്റ് ആവശ്യകതകൾ എന്നിവ നോക്കാം.

### openai ഇൻസ്റ്റാൾ ചെയ്യുക

  > [!NOTE] ഈ നോട്ട്ബുക്ക് Codespaces-ൽ അല്ലെങ്കിൽ Devcontainer-ൽ പ്രവർത്തിപ്പിക്കുന്നില്ലെങ്കിൽ മാത്രമേ ഈ ഘട്ടം ആവശ്യമായിരിക്കൂ


OpenAI അല്ലെങ്കിൽ Azure OpenAI-യുമായി ഇടപഴകാൻ നിരവധി ലൈബ്രറികൾ ഉണ്ട്. C#, Python, JavaScript, Java തുടങ്ങിയ പല പ്രോഗ്രാമിംഗ് ഭാഷകളും ഉപയോഗിക്കാം.  
നാം `openai` Python ലൈബ്രറി ഉപയോഗിക്കാൻ തീരുമാനിച്ചു, അതിനാൽ `pip` ഉപയോഗിച്ച് ഇൻസ്റ്റാൾ ചെയ്യാം.

```bash
pip install openai
```

നിങ്ങൾ ഈ നോട്ട്ബുക്ക് Codespaces-ൽ അല്ലെങ്കിൽ Dev Container-ൽ പ്രവർത്തിപ്പിക്കുന്നില്ലെങ്കിൽ, നിങ്ങളുടെ യന്ത്രത്തിൽ [Python](https://www.python.org/) ഇൻസ്റ്റാൾ ചെയ്യേണ്ടതുണ്ട്.

### ഒരു റിസോഴ്‌സ് സൃഷ്ടിച്ച് നിങ്ങളുടെ API കീ കണ്ടെത്തുക

ഇതുവരെ ചെയ്തിട്ടില്ലെങ്കിൽ, താഴെപ്പറയുന്ന ഘട്ടങ്ങൾ ചെയ്യേണ്ടതാണ്:

- OpenAI-യിൽ ഒരു അക്കൗണ്ട് സൃഷ്ടിക്കുക <https://platform.openai.com/signup>.
- ഇപ്പോൾ നിങ്ങളുടെ API കീകൾ നേടുക <https://platform.openai.com/api-keys>. 

>[!NOTE]
> നിങ്ങളുടെ API കീ കോഡിൽ നിന്ന് വേർതിരിക്കുന്നത് നല്ലതാണ്. ഇത് പരിസ്ഥിതി ചാരങ്ങളിലൂടെ ചെയ്യാം.
> - നിങ്ങളുടെ .env ഫയലിൽ `OPENAI_KEY` എന്ന പരിസ്ഥിതി ചാരത്തെ നിങ്ങളുടെ API കീ ആയി സജ്ജമാക്കുക. ഈ കോഴ്സിന്റെ മുൻപത്തെ അഭ്യാസങ്ങൾ നിങ്ങൾ പൂർത്തിയാക്കിയിട്ടുണ്ടെങ്കിൽ, നിങ്ങൾ സജ്ജമാണ്.
> - API കീ ഒരിക്കൽ മാത്രമേ ലഭ്യമാകൂ എന്നത് ശ്രദ്ധിക്കുക. അതിനാൽ ശരിയായി പകർത്തിയിട്ടുണ്ടെന്ന് ഉറപ്പാക്കുക. പ്രവർത്തിക്കാത്ത പക്ഷം കീ നീക്കം ചെയ്ത് പുതിയതായി സൃഷ്ടിക്കുക.

### OpenAI കോൺഫിഗറേഷൻ സജ്ജമാക്കുക

OpenAI ഉപയോഗിക്കുന്നുവെങ്കിൽ, കോൺഫിഗറേഷൻ ഇങ്ങനെ സജ്ജമാക്കാം:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )
```

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-3.5-turbo"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"
messages = [{"role": "user", "content": prompt}]  

# make completion
completion = client.chat.completions.create(model=deployment, messages=messages, max_tokens=600)

# print response
print(completion.choices[0].message.content)
````
