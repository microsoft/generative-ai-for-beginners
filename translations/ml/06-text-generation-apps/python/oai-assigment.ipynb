# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്ലിക്കേഷനുകൾ നിർമ്മിക്കുക

ഇതുവരെ നിങ്ങൾ ഈ പാഠ്യപദ്ധതിയിൽ കണ്ടിട്ടുള്ളത് പോലെ പ്രോപ്റ്റുകൾ പോലുള്ള പ്രാഥമിക ആശയങ്ങളും "പ്രൊംപ്റ്റ് എന്റർപ്രണറിംഗ്" എന്ന് വിളിക്കുന്ന പൂർണ്ണ വിഷയവുമുണ്ട്. ChatGPT, Office 365, Microsoft Power Platform എന്നിവ പോലുള്ള നിരവധി ടൂളുകൾ നിങ്ങൾ പ്രോപ്റ്റുകൾ ഉപയോഗിച്ച് ഏതെങ്കിലും ഒന്നും നേടാൻ സഹായിക്കുന്നു.

നിങ്ങൾക്ക് ആപ്ലിക്കേഷനിൽ ഇങ്ങനെ ഒരു അനുഭവം കൂട്ടാൻ ആഗ്രഹമാണെങ്കിൽ, പ്രോപ്റ്റുകൾ, പൂർത്തീകരണങ്ങൾ തുടങ്ങിയ ആശയങ്ങൾ മനസ്സിലാക്കണം, കൂടാതെ ഉപയോഗിക്കാൻ ലൈബ്രറി തിരഞ്ഞെടുക്കണം. ഇതാണ് ഈ അധ്യായത്തിൽ നിങ്ങൾ പഠിക്കാനിരിക്കുന്നത്.

## പരിചയം

ഈ അധ്യായത്തിൽ നിങ്ങൾ:

- openai ലൈബ്രറിയും അതിന്റെ പ്രാഥമിക ആശയങ്ങളും പഠിക്കും.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കും.
- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ പ്രോപ്റ്റ്, താപനില, ടോക്കണുകൾ പോലുള്ള ആശയങ്ങൾ എങ്ങനെ ഉപയോഗിക്കാമെന്ന് മനസ്സിലാക്കും.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠത്തിന്റെ അവസാനം, നിങ്ങൾക്കാകും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- നിങ്ങളുടെ ആപ്പ് കുറവോ അധികമോ ടോക്കണുകൾ ഉപയോഗിക്കാൻ കൂടാതെ താപനില മാറ്റി വിവിധ ഔട്ട്പുട്ടുകൾ സൃഷ്ടിക്കാൻ ക്രമീകരിക്കുക.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന്?

സാധാരണയായി നിങ്ങൾ ഒരു ആപ്ലിക്കേഷൻ നിർമ്മിക്കുമ്പോൾ അത് ഇനിപ്പറയുന്ന പോലുള്ള ഏതെങ്കിലും ഇന്റർഫേസ് ഉണ്ടാകുന്നു:

- കമാൻഡ് അധിഷ്ഠിതം. കൺസോൾ ആപ്ലിക്കേഷനുകൾ നാമം പറയുന്നതുപോലെ കമാൻഡ് ടൈപ്പ് ചെയ്യുകയും അത് ആശയമോ പ്രവർത്തനമോ നടത്തുകയും ചെയ്യുന്നു. ഉദാഹരണത്തിന്, `git` എന്നത് ഒരു കമാൻഡ് അധിഷ്ഠിത ആപ്പ് ആണ്.
- ഉപയോക്തൃ ഇന്റർഫേസ് (UI). ചില ആപ്പുകളിൽ ഗ്രാഫിക്കൽ ഉപയോക്തൃ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ട്, അവയിൽ ബട്ടണുകൾ ക്ലിക്ക് ചെയ്യുക, ടെക്സ്റ്റ് നൽകുക, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കുക തുടങ്ങിയവ ചെയ്യാം.

### കൺസോൾ ആപ്പുകളും UI ആപ്പുകളും പരിമിതമാണ്

നിങ്ങൾ ടൈപ്പ് ചെയ്യുന്ന കമാൻഡ് ക്രമീകരിക്കപ്പെട്ട ആപ്പിന് താരതമ്യം ചെയ്യൂ: 

- **ഇത് പരിധിയും അനുവദനീയതകളുംകൂടിയതാണ്**. നിങ്ങൾക്ക് ഏതെങ്കിലും കമാൻഡ് ടൈപ്പിക്കാൻ സാധിക്കില്ല, കൂടാതെ ആപ്പ് അനുവദിക്കുന്നവ മാത്രമേ ഇത് പിന്തുണയ്ക്കൂ.
- **ഭാഷാ പ്രത്യേകം**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കാം, എന്നാൽ ഡിഫോള്റ്റായി ആപ്പ് നിർദ്ദിഷ്ട ഒരു ഭാഷയ്ക്കായി നിർമ്മിക്കപ്പെടുന്നു, കൂടുതൽ ഭാഷ പിന്തുണ ചേർക്കാം എന്നാലും.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളുടെ ഗുണങ്ങൾ

അങ്ങനെയായാൽ, ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എങ്ങനെ വ്യത്യസ്തമാണ്?

ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളിൽ നിങ്ങൾക്ക് കൂടുതൽ സ്വതന്ത്രതകൾ ഉണ്ട്, നിങ്ങൾ കമാൻഡുകളോ പ്രത്യേക ഇൻപുട്ട് ഭാഷയിലോ പരിമിതരല്ല. പകരം, നിങ്ങൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ഇടപഴകാം. മറ്റൊരു ഗുണം ആയിരിക്കുന്നത്, നിങ്ങൾ വലിയ മാഹിതികളുടെ ശേഖരത്തിൽ പരിശീലിപ്പിച്ച ഒരു ഡാറ്റാ സ്രോതസ്സുമായി ഇതിനകം ഇടപഴകുന്നുവെന്ന് കൊണ്ടാണ്, പരമ്പരാഗത ആപ്പുകൾ ഒരു ഡാറ്റാബേസിലുള്ള വിവരങ്ങൾ മാത്രം പരിമിതപ്പെട്ടിരിക്കുന്നു.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ ഞാൻ എന്തെന്തു നിർമ്മിക്കാം?

നിങ്ങൾക്ക് നിരവധി കാര്യങ്ങൾ നിർമ്മിക്കാനാകും. ഉദാഹരണങ്ങൾ:

- **ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനി, അതിന്റെ ഉൽപ്പന്നങ്ങൾ എന്നിവയെപ്പറ്റി ചോദ്യങ്ങൾക്ക് ഉത്തരം നൽകുന്ന ചാറ്റ്ബോട്ട് നല്ല അനുയോജ്യമായിരിക്കും.
- **സഹായകൻ**. LLMകൾ ടെക്സ്റ്റ് സംക്ഷേപിക്കുക, ടെക്സ്റ്റിൽ നിന്നുള്ള അറിവുകൾ എടുക്കുക, റിസ്യൂമുകൾ പോലുള്ള ടെക്സ്റ്റ് നിർമ്മിക്കുക തുടങ്ങിയ കാര്യങ്ങളിൽ നല്ല പ്രകടനം കാണിക്കുന്നു.
- **കോഡ് അസിസ്റ്റന്റ്**. നിങ്ങൾ ഉപയോഗിക്കുന്ന ഭാഷാ മോഡലിന്റെ അടിസ്ഥാനത്തിൽ, നിങ്ങൾക്ക് കോഡ് എഴുതുന്നതിനായി സഹായിക്കുന്ന ഒരു കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാമാകാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉൽപ്പന്നം ഉപയോഗിക്കാം, അതുപോലെ ChatGPT സഹായിക്കും.

## എങ്ങനെ തുടങ്ങാം?

സാധാരണയായി, LLM-നുമായി സംയോജിപ്പിക്കുന്നതിന് താഴെ പറയുന്ന രണ്ട് സമീപനങ്ങളിലൊരു ഒന്ന് സ്വീകരിക്കണം:

- API ഉപയോഗിക്കുന്നത്. ഇവിടെ നിങ്ങൾ പ്രോംപ്റ്റ് ഉപയോഗിച്ച് വെബ് അഭ്യർത്ഥനകൾ സൃഷ്ടിക്കുകയും, അവയിൽ നിന്നുള്ള ടെക്സ്റ്റ് തിരിച്ചുപിടിക്കുകയും ചെയ്യുന്നു.
- ലൈബ്രറി ഉപയോഗിക്കൽ. ലൈബ്രറികൾ API കാൾസ് എൻക്യാപ്സുലേറ്റ് ചെയ്ത് ഉപയോഗിക്കാൻ എളുപ്പമാകും.

## ലൈബ്രറികൾ / SDKകൾ

LLM ഉപയോഗിക്കുന്നതിന് ചില പ്രശസ്ത ലൈബ്രറികൾ:

- **openai**, ഈ ലൈബ്രറി നിങ്ങളുടെ മോഡലുമായി ബന്ധിപ്പിക്കാൻ, പ്രോംപ്റ്റുകൾ അയയ്ക്കാൻ സഹായിക്കുന്നു.

അവയോടൊപ്പം ഉയർന്ന തലത്തിൽ പ്രവർത്തിക്കുന്ന ലൈബ്രറികളുമുണ്ട്:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python പിന്തുണക്കുന്നു.
- **Semantic Kernel**. Semantic Kernel മൈക്രോസോഫ്റ്റിന്റെ ഒരു ലൈബ്രറിയാണ്, C#, Python, Java ഭാഷകൾ പിന്തുണയ്ക്കുന്നു.

## openai ഉപയോഗിച്ച് ആദ്യ ആപ്പ്

ഇപ്പോൾ നമ്മൾ എങ്ങനെ ആദ്യ ആപ്പ് നിർമ്മിക്കാമെന്നു നോക്കാം, എത്ര ലൈബ്രറികൾ വേണം, എത്ര ആവശ്യമുണ്ട് മുതലായവ.

### openai ഇൻസ്റ്റാൾ ചെയ്യുക

  > [!NOTE] ഈ പടി Codespaces-ൽ അല്ലെങ്കിൽ Devcontainer-ൽ പ്രവർത്തിപ്പിക്കുന്ന പക്ഷം ആവശ്യമായില്ല


OpenAI അല്ലെങ്കിൽ Azure OpenAI യുമായി ഇടപഴകാൻ നിരവധി ലൈബ്രറികൾ ലഭ്യമാണ്. C#, Python, JavaScript, Java തുടങ്ങിയ പല പ്രോഗ്രാമിംഗ് ഭാഷകൾ ഉപയോഗിക്കാം.  
നാം `openai` Python ലൈബ്രറി തിരഞ്ഞെടുക്കുകയോ, അതിനാൽ `pip` ഉപയോഗിച്ച് ഇൻസ്റ്റാൾ ചെയ്യുക.

```bash
pip install openai
```

നിങ്ങൾ ഈ നോട്ട്ബുക്ക് Codespaces-ൽ അല്ലെങ്കിൽ Dev Container-ൽ പ്രവർത്തിപ്പിക്കുന്നില്ലെങ്കിൽ, നിങ്ങളുടെ യന്ത്രത്തിൽ [Python](https://www.python.org/) ഇൻസ്റ്റാൾ ചെയ്യേണ്ടതാണ്.

### ഒരു റിസോഴ്‌സ് സൃഷ്ടിക്കുക, API കീ കണ്ടെത്തുക

നിങ്ങൾ ഇതുവരെ ചെയ്തില്ലെങ്കിൽ, താഴെ പറയുന്ന ക്രമങ്ങൾ നിർവ്വഹിക്കണം:

- OpenAI-യിൽ ഒരു അക്കൗണ്ട് സൃഷ്ടിക്കുക <https://platform.openai.com/signup>.
- ഇതിനു ശേഷം, നിങ്ങളുടെ API കീകൾ നേടുക <https://platform.openai.com/api-keys>. 

>[!NOTE]
> നിങ്ങളുടെ API കീ നിങ്ങളുടെ കോഡിൽ നിന്ന് വേർതിരിക്കാൻ നല്ലതു. ഇത് എങ്ങനെ ചെയ്യാം എന്ന് കാണുക.
> - നിങ്ങളുടെ .env ഫയലിൽ `OPENAI_KEY` എന്ന എൻവയോൺമെന്റ് വേരിയബിൾ നിങ്ങളുടെ API കിയാക്കി സജ്ജമാക്കുക. ഇതിനകം ഈ കോഴ്സിലെ മുന്‍ പരിശീലനങ്ങൾ പൂർത്തിയായെങ്കിൽ നിങ്ങൾ كله സജ്ജമാണ്.
> - API കീ ഒരിക്കൽ മാത്രം പ്രാപ്യമായിരിക്കും. അതിനാൽ, ശരിയായി കോപ്പി ചെയ്തിട്ടുണ്ടോ എന്ന് സ്ഥിരീകരിക്കുക അതു പ്രവർത്തിക്കുന്നില്ലെങ്കിൽ കടുത്ത വശം കDeletion ആക്കും, പുതിയ കീ സൃഷ്ടിക്കുമെന്നു ശുപാർശ ചെയ്യപ്പെടുന്നു.


### OpenAI ക്രമീകരണം സജ്ജമാക്കുക

OpenAI ഉപയോഗിക്കുന്നപക്ഷം, ക്രമീകരണം ഇങ്ങനെ ചെയ്യാം:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"
```

മേലുപറഞ്ഞ രീതിയിൽ ക്രമീകരിക്കുന്നു:

- `api_key`, OpenAI ഡാഷ്‌ബോർഡിൽനിന്നുള്ള вашей API കീ ആണ്.
- `deployment`, നിങ്ങളുടെ GPT പതിപ്പ്.

> [!NOTE]
> `os.environ` എന്നത് എൻവയോൺമെന്റ് വേരിയബിൾ വായിക്കുന്ന ഫംഗ്ഷൻ ആണ്. ഇത് ഉപയോഗിച്ച് `OPENAI_API_KEY` പോലുള്ള എൻവയോൺമെന്റ് വേരിയബിളുകൾ വായിക്കാം.

## ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യുക

ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യാനുള്ള വഴി `responses.create` മെത്തേഡ് ഉപയോഗിക്കുകയാണ്. ഉദാഹരണം ഇവിടെ:

```python
prompt = "Complete the following: Once upon a time there was a"

print(response.output_text)
```

മേലിലുള്ള കോഡിൽ, ഒന്നാം മത് ഒരു റസ്പോൺസ് ഒബ്‌ജക്റ്റ് സൃഷ്ടിച്ച്, ഞങ്ങൾ ഉപയോഗിക്കാൻ ഉദ്ദേശിക്കുന്ന മോഡലും പ്രോപ്പ്റ്റും പാസാക്കുന്നു. പിന്നീട് ജനറേറ്റ് ചെയ്ത ടെക്സ്റ്റ് പ്രിന്റ് ചെയ്യുന്നു.

### ചാറ്റ് റസ്പോൺസുകൾ

ഇതുവരെ നിങ്ങൾ ഒരു പ്രോപ്പ്റ്റിൽ നിന്ന് എങ്ങനെ ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യാമെന്ന് കണ്ടു. Responses API ചാറ്റ്ബോട്ടുകൾക്കും അനുയോജ്യമാണ് - നിങ്ങൾ സന്ദേശങ്ങളുടെ ഒരു ലിസ്റ്റ് `input` ആയി നൽകി സംഭാഷണം സൃഷ്ടിക്കാം. ഉദാഹരണമായി ഇതു കാണുക:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

ഈ ഫങ്ഷനാലിറ്റി അടുത്ത അധ്യായത്തിൽ കൂടുതൽ വിശദീകരിക്കും.

## വ്യത്യസ്ത കാര്യങ്ങൾക്ക് വ്യത്യസ്ത തരത്തിലുള്ള പ്രോംപ്റ്റുകൾ

ഇപ്പോൾ നിങ്ങൾ പ്രോംപ്റ്റ് ഉപയോഗിച്ച് എങ്ങനെ വാചകം ഉണ്ടാക്കാമെന്ന് കണ്ടു. വിവിധ തരത്തിലുള്ള വാചകങ്ങൾ സൃഷ്ടിക്കാൻ നിങ്ങൾക്ക് മാറ്റം വരുത്താനും മാറ്റാനും സാധിക്കുന്ന ഒരു പ്രോഗ്രാം വരെ നിങ്ങൾക്കുണ്ട്.

- **ഒരു തരത്തിലുള്ള വാചകം സൃഷ്ടിക്കുക**. ഉദാഹരണത്തിന്, നിങ്ങള്‍ക്ക് ഒരു കവിത, ഒരു ക്വിസിനുള്ള ചോദ്യങ്ങൾ തുടങ്ങിയവ സൃഷ്ടിക്കാം.

- **കോഡ് സൃഷ്ടിക്കുക**. ഇമെയിൽ സാധുത പരിശോദിക്കാൻ ഉപയോഗിക്കുന്ന നിബന്ധനാ വ്യാകരണത്തിന് പോലെയുള്ള കോഡ് സൃഷ്ടിക്കാൻ പ്രോംപ്റ്റുകൾ ഉപയോഗിക്കാം, അല്ലെങ്കിൽ ഒരു വെബ് ആപ്പ് പോലുള്ള മുഴുവൻ പ്രോഗ്രാമും സൃഷ്ടിക്കാം.

1. നിലവിലുള്ള *app.py* ഫയൽ ആരംഭബിന്ദുവായി ഉപയോഗിക്കുക
1. `prompt` വ്യത്യസ്തം കണ്ടെത്തി അതിന്റെ കോട് താഴെപ്പറയുന്നവ ആയി മാറ്റുക:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

   > ശ്രദ്ധിക്കുക, 1.0 നികടത്തോളം വരുന്നത് ഔട്ട്പുട്ടിൽ കൂടുതൽ വൈവിധ്യം ഉണ്ടാക്കും.
