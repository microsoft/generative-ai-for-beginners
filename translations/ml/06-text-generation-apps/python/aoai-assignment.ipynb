# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾ തയാറാക്കുക

ഈ പാഠപദ്ധതിയിൽ ഇതുവരെ നിങ്ങൾ കണ്ടിരിക്കുന്നത് പ്രോംപ്റ്റുകൾ പോലുള്ള മുള പ്രധാന ആശയങ്ങളാണ്, ഇത്തരത്തിലൊരു ശാഖയെ "പ്രോംപ്റ്റ് എഞ്ചിനീയറിങ്" എന്നും വിളിക്കുന്നു. ChatGPT, Office 365, Microsoft Power Platform തുടങ്ങിയ പല ഉപകരണങ്ങളും പ്രോംപ്റ്റുകൾ ഉപയോഗിച്ച് ഒന്നൊന്നുകിൽ പ്രവർത്തനങ്ങൾ സാധ്യമാക്കാൻ സഹായിക്കുന്നു.

ഒരു ആപ്പിൽ ഇത്തരമുള്ള അനുഭവം ചേർക്കാൻ, പ്രോംപ്റ്റുകൾ, കോംപ്ലീഷനുകൾ പോലുള്ള ആശയങ്ങൾ മനസിലാക്കുകയും അനുയോജ്യമായ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കുകയും വേണം. ഇതാണ് നിങ്ങൾ ഈ അദ്ധ്യായത്തിൽ പഠിക്കാൻ പോകുന്നത്.

## പരിചയം

ഈ അദ്ധ്യായത്തിൽ, നിങ്ങൾ:

- ഓപ്പൺഎഐ ലൈബ്രറിയും അതിന്റെ മുള മുഖ്യ ആശയങ്ങളും പഠിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുന്നതിനായി പ്രോംപ്റ്റ്, ടെംപറേച്ചർ, ടോക്കൺസ് എന്നിവ ഉപയോഗിക്കുന്നത് മനസിലാക്കുക.

## പഠനലക്ഷ്യങ്ങൾ

ഈ പാഠം അവസാനിപ്പിച്ചാൽ, നിങ്ങൾക്ക് കഴിയും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- കൂടുതൽ അല്ലെങ്കിൽ കുറച്ചെങ്കിലും ടോക്കൺസ് ഉപയോഗിച്ച്, കൂടാതെ വ്യത്യസ്ത ഫലങ്ങൾക്ക് ടെംപറേച്ചർ മാറ്റി ആപ്പ് ക്രമീകരിക്കുക.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി ഒരു ആപ്പ് നിർമ്മിക്കുമ്പോൾ താഴെ കാണുന്ന തരത്തിലുള്ള ഇൻറർഫേസ് ഉണ്ടാകാം:

- കമാൻഡ് അടിസ്ഥാനമാണ്. കൺസോൾ ആപ്പുകൾ സാധാരണയായി കമ്മാൻഡ് ടൈപ്പ് ചെയ്ത് ഒരു ജോലി നിർവ്വഹിക്കുന്നവയാണ്. ഉദാഹരണത്തിന്, `git` ഒരു കമാൻഡ് അടിസ്ഥാന ആപ്പ് ആണ്.
- ഉപയോക്തൃ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾക്ക് ഗ്രാഫിക്കൽ യൂസർ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ടാകും, അവിടെ നിങ്ങൾ ബട്ടണുകൾ ക്ലിക്ക് ചെയ്യുകയും, ടെക്സ്റ്റ് ഇൻപുട്ട് നൽകുകയും, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കുകയും ചെയ്യാം.

### കൺസോൾ, UI ആപ്പുകൾക്ക് പരിമിതികളുണ്ട്

കമാൻഡ് ടൈപ്പ് ചെയ്യുന്ന കമാൻഡ് ആപ്പിനോട് താരതമ്യം ചെയ്യുക:

- **ഇത് പരിധിയുള്ളതാണ്**. നിങ്ങൾക്ക് എവിടെയും കമാൻഡ് ടൈപ്പ് ചെയ്യാൻ കഴിയില്ല, ആപ്പിന് പിന്തുണയുള്ള കമാൻഡുകൾ മാത്രം ഉപയോഗിക്കാം.
- **ഭാഷാ പ്രത്യേകമായത്**. ചില ആപ്പുകൾ നിരവധി ഭാഷകൾ പിന്തുണയ്ക്കാം, എന്നാൽ സാധാരണയായി ആപ്പ് ഒരു പ്രത്യേക ഭാഷയ്ക്കായി രൂപകൽപ്പന ചെയ്തതാണ്, കൂടുതൽ ഭാഷകൾ ചേർക്കാൻ സാധിക്കാം.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളുടെ ഗുണങ്ങൾ

ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എനിക്ക് എങ്ങനെ വ്യത്യാസമാണെന്ന്?

ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ, നിങ്ങൾക്ക് കൂടുതൽ സ്വാതന്ത്ര്യമുണ്ട്, നിങ്ങൾ കമ്മാൻഡുകൾക്കോ ഒരു പ്രത്യേക ഇൻപുട്ട് ഭാഷയ്ക്കോ പരിമിതരാക്കപ്പെട്ടിട്ടില്ല. മറിച്ച്, നിങ്ങൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി സംവദിക്കാം. മറ്റൊരു ഗുണം, നിങ്ങൾ ഇതിനകം വലിയ പാര്‍പ്പുസ്സിൽ പരിശീലനം നേടിയ ഡാറ്റയാണ് ഉപയോഗിക്കുന്നത്, സാധാരണ ആപ്പ് ഡാറ്റാബേസിൽ ഉള്ളതിലേക്കാൾ കൂടുതലാണ്.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ ഞാൻ എന്ത് നിർമ്മിക്കാമെന്നും?

നിങ്ങൾക്ക് നിരവധി കാര്യങ്ങൾ നിർമ്മിക്കാനാകും. ഉദാഹരണത്തിന്:

- **ചാറ്റ് ബോട്ട്**. നിങ്ങളുടെ കമ്പനി, ഉൽപ്പന്നങ്ങൾ എന്നിവയെക്കുറിച്ച് ചോദ്യങ്ങൾ ഉത്തരം നൽകുന്ന ചാറ്റ്ബോട്ട് നല്ലൊരു തിരഞ്ഞെടുപ്പാകാം.
- **ഹelper**. LLM-കൾ ടെക്സ്റ്റ് സംഗ്രഹിക്കാനും, ടെക്സ്റ്റിൽ നിന്നുള്ള洞察ങ്ങൾ എടുക്കാനും, പുനരാരംഭം പോലുള്ള ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യാനും യോഗ്യമാണ്.
- **കോടിംഗ് സഹായി**. നിങ്ങൾ 사용하는 ഭാഷാ മോഡലിനെ ആശ്രയിച്ച്, കോഡ് എഴുതാൻ സഹായിക്കുന്ന ഒരു കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന് GitHub Copilot പോലുള്ള ഉൽപ്പന്നങ്ങളോ ChatGPT-ഉം ഉപയോഗിച്ച് കോഡ് എഴുതാൻ സഹായിക്കാം.

## എങ്ങനെ ആരംഭിക്കാം?

സാധാരണയായി, നിങ്ങൾക്ക് LLM-ഉം ഇന്റഗ്രേറ്റ് ചെയ്യാനുള്ള മാർഗ്ഗങ്ങൾ രണ്ട് വിധങ്ങൾ ഉള്ളതാണ്:

- API ഉപയോഗിക്കുക. ഇവിടെ നിങ്ങൾ നിങ്ങളുടെ പ്രോംപ്റ്റ് ഉപയോഗിച്ച് വെബ് അഭ്യർത്ഥനകൾ നിർമ്മിച്ച്, ജനറേറ്റ് ചെയ്ത ടെക്സ്റ്റ് തിരിച്ചുപിടിക്കും.
- ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API വിളിക്കലുകൾ ഉൾക്കൊള്ളിക്കുകയും ഉപയോഗിക്കാൻ എളുപ്പമാക്കുകയും ചെയ്യും.

## ലൈബ്രറികൾ/SDKകൾ

LLM-കൾകൊണ്ട് പ്രവർത്തിക്കാൻ അറിയപ്പെടുന്ന ചില ലൈബ്രറികൾ:

- **openai**, ഈ ലൈബ്രറി നിങ്ങളുടെ മോഡലുമായി ബന്ധപ്പെടാനും പ്രോംപ്റ്റുകൾ അയക്കാനും സഹായിക്കുന്നു.

പിന്നെ ഉയർന്ന മടിയിൽ പ്രവർത്തിക്കുന്ന ലൈബ്രറികൾ:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python-നെ പിന്തുണക്കുന്നു.
- **Semantic Kernel**. സെമാന്റിക് കർണൽ മൈക്രോസോഫ്റ്റിന്റെ ലൈബ്രറി ആണ്, C#, Python, Java ഭാഷകൾ പിന്തുണയ്ക്കുന്നു.

## openai ഉപയോഗിച്ച് ആദ്യ ആപ്പ്

നിങ്ങളുടെ ആദ്യ ആപ്പ് എങ്ങനെയാണെന്നു നോക്കാം, എത്ര ലൈബ്രറികൾ വേണം, എത്ര ആവശ്യമാണ് എന്നതു തുടങ്ങിയവ.

### openai ഇൻസ്റ്റാൾ ചെയ്യുക

  > [!NOTE] നിങ്ങൾ ഈ നോട്ട്ബുക്ക് Codespaces অথবা Devcontainer-ൽ പ്രവർത്തിപ്പിക്കുന്നുവെങ്കിൽ ഈ ഘട്ടം ആവശ്യമില്ല.


OpenAI അല്ലെങ്കിൽ Azure OpenAI-യുമായി സംവദിക്കാൻ നിരവധി ലൈബ്രറികൾ ലഭ്യമാണ്. C#, Python, JavaScript, Java തുടങ്ങിയ പല പ്രോഗ്രാമിംഗ് ഭാഷകളും ഉപയോഗിക്കാം.  
നമുക്ക് `openai` എന്ന Python ലൈബ്രറി ഉപയോഗിക്കാനാണ് തീരുമാനിച്ചത്, അതിനാൽ `pip` ഉപയോഗിച്ച് ഇൻസ്റ്റാൾ ചെയ്യാം.

```bash
pip install openai
```

Codespaces അല്ലെങ്കിൽ Dev Container-ൽ ഈ നോട്ട്ബുക്ക് പ്രവർത്തിപ്പിക്കാത്തപക്ഷം, നിങ്ങളുടെ കമ്പ്യൂട്ടറിൽ [Python](https://www.python.org/) ഇൻസ്റ്റാൾ ചെയ്യേണ്ടതുണ്ട്.

### റിസോഴ്‌സ് സൃഷ്ടിക്കുക

ഇപ്പോൾ വരെ ചെയ്തിട്ടില്ലെങ്കിൽ, താഴെപ്പറയുന്ന ഘട്ടങ്ങൾ പാലിക്കണം:

- Azure-ൽ അക്കൗണ്ട് നിർമ്മിക്കുക <https://azure.microsoft.com/free/>.
- Microsoft Foundryയിൽ [Azure OpenAI റിസോഴ്‌സ്](https://ai.azure.com?WT.mc_id=academic-105485-koreyst) സൃഷ്ടിക്കുക. [ഉൾപ്പെടുത്തൽ ഗൈഡ്](https://learn.microsoft.com/azure/ai-services/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) കാണുക.


### API കീയും എന്റ്പോയിന്റും കണ്ടെത്തുക

ഇനി നിങ്ങളുടെ `openai` ലൈബ്രറിയെ ഉപയോഗിക്കാനായി API കീ എന്താണെന്ന് അറിയിക്കണം. API കീ കണ്ടെത്താൻ, Azure OpenAI റിസോഴ്‌സിന്റെ "Keys and Endpoint" ഭാഗത്ത് പോയി "Key 1" മൂല്യം കാപി ചെയ്യുക.

  ![Keys and Endpoint resource blade in Azure Portal](https://learn.microsoft.com/azure/ai-services/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ഇപ്പോൾ അതുകൊണ്ടുള്ള വിവരങ്ങൾ കോപ്പി ചെയ്‌തതിനുശേഷം, ലൈബ്രറിക്ക് അത് ഉപയോഗിക്കാൻ ആവശ്യമായ നിർദ്ദേശങ്ങൾ നൽകുക.

> [!NOTE]
> നിങ്ങളുടെ API കീ നിങ്ങളുടെ കോഡിൽ നിന്ന് വേർതിരിക്കേണ്ടത് ഉത്തമമാണ്. അതിനായി എൻവയോൺമെന്റ് വേരിയബിളുകൾ ഉപയോഗിക്കാം.
> - നിങ്ങളുടെ .env ഫയലിൽ `AZURE_OPENAI_API_KEY` എന്ന എൻവയോൺമെന്റ് വാരിയബിള് API കീ ആയി ക്രമീകരിക്കുക. ഈ പാഠപദ്ധതി മുൻപ് നടത്തിയ യഥാർത്ഥ അഭ്യാസങ്ങൾ ഇതുവരെ ചെയ്തിട്ടുണ്ടെങ്കിൽ, നിങ്ങൾ ഇതിനകം തയ്യാറാണ്.


### Azure കോൺഫിഗറേഷൻ സജ്ജമാക്കുക

നിങ്ങൾ Azure OpenAI ഉപയോഗിക്കുന്നുവെങ്കിൽ കോൺഫിഗറേഷൻ ഇങ്ങനെ സജ്ജമാക്കാം. Responses API Azure OpenAI **v1 endpoint** നിന്നാണ് സേവനം നൽകുന്നത്, അതിനാൽ `OpenAI` ക്ളയന്റിന് `<your-endpoint>/openai/v1/`-ൽ പോയിന്റ് ചെയ്യുക:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

മേലിലുള്ള കോഡിൽ നമുക്ക് ഇവ ക്രമീകരിക്കുന്നു:

- `api_key`, ഇത് Azure പോർടലിൽ നിന്നുള്ള നിന്ന API കീ ആണ്.
- `base_url`, ഇത് Azure OpenAI എന്റ്പോയിന്റ് ആണ് `/openai/v1/` ചേർത്തത്. API ക്കർവേർഷൻ ഇനി നൽകേണ്ട ആവശ്യമില്ല.
- `deployment`, ഫൗണ്ട്രി പോർട്ടലിൽ നിങ്ങൾ സൃഷ്ടിച്ച മോഡൽ ഡിപ്ലോയ്‌മെൻറ് പേര്.

> [!NOTE]
> `os.environ` എന്നത് എൻവയോൺമെന്റ് വേരിയബിളുകൾ വായിക്കാൻ ഉപയോഗിക്കുന്ന മാപ്പിംഗ് ആണ്. നിങ്ങൾ ഇത് ഉപയോഗിച്ച് `AZURE_OPENAI_API_KEY` , `AZURE_OPENAI_ENDPOINT` പോലുള്ള എൻവയോൺമെന്റ് വേരിയബിളുകൾക്ക് ആക്‌സസ് പെടാം.

```python
prompt = "Complete the following: Once upon a time there was a"
```

### ചാറ്റ് മറുപടികൾ

Responses API ഏകഗ്രവണ ടെക്സ്റ്റ് ജനറേഷന് മാത്രമല്ല, മൾട്ടി ടേൺ ചാറ്റ്ബോട്ട് നിർമാണത്തിനും അനുയോജ്യമാണ് — `input` ലിസ്റ്റിൽ കൂടുതൽ സന്ദേശങ്ങൾ നൽകികെ ചർച്ച നിർമ്മിക്കാം:

```python

  )
```

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?
