# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾ നിർമ്മിക്കുക

ഈ പാഠത്തിൽ ഇതുവരെ നിങ്ങൾ കണ്ടത് പോലെ പ്രോംപ്റ്റുകൾ പോലുള്ള കോർ ആശയങ്ങൾ ഉണ്ട്, "പ്രോംപ്റ്റ് എന്‍ജിനീയറിങ്" എന്നു പേരുള്ള ഒരു പൂർണ്ണ വിഭാഗവും ഉണ്ട്. ChatGPT, Office 365, Microsoft Power Platform പോലുള്ള പല ടൂളുകളും പ്രോംപ്റ്റുകൾ ഉപയോഗിച്ച് ഒരുപാട് കാര്യങ്ങൾ കൈകാര്യം ചെയ്യാൻ നിങ്ങളെ പിന്തുണയ്ക്കുന്നു.

ഒരു ആപ്പിൽ ഇത്തരത്തിലുള്ള ഒരു അനുഭവം ചേർക്കാൻ, പ്രോംപ്റ്റുകൾ, പൂർത്തീകരണങ്ങൾ എന്നിവ പോലുള്ള ആശയങ്ങൾ മനസ്സിലാക്കണം കൂടാതെ ഉപയോഗിക്കാൻ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കണം. നിങ്ങൾ ഈ അധ്യായത്തിൽ അതെല്ലാം പഠിക്കും.

## പരിചയം

ഈ അധ്യായത്തിൽ നിങ്ങൾ ചെയ്യും:

- openai ലൈബ്രറിയും അതിന്റെ കോർ ആശയങ്ങളും പഠിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ പ്രോംപ്റ്റ്, ടേംപറേച്ചർ, ടോക്കൺ എന്നിവ ഉപയോഗിക്കുന്ന വിധം മനസ്സിലാക്കുക.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠത്തിന്റെ അവസാനം, നിങ്ങൾക്ക് കഴിയും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്തെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- നിങ്ങളുടെ ആപ്പിനെ കൂടുതൽ അല്ലെങ്കിൽ കുറവായ ടോക്കണുകൾ ഉപയോഗിക്കാൻ കോൺഫിഗർ ചെയ്തേക്കാം, കൂടാതെ ഫലം വൈവిధ്യമാക്കാൻ ടേംപറേച്ചർ മാറ്റാം.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി നിങ്ങൾ ഒരു ആപ്പ് നിർമ്മിക്കുമ്പോൾ ഇതുപോലുള്ള ഏതെങ്കിലും ഇന്റർഫേസ് ഉണ്ടായിരിക്കും:

- കമാൻഡ് അടിസ്ഥാനമാക്കി. കോൺസോൾ ആപ്പുകൾ സാധാരണ ആപ്പുകളായി കമാൻഡ് ടൈപ്പ് ചെയ്തു ടാസ്ക് നടത്തുന്നതാണ്. ഉദാഹരണത്തിന്, `git` ഒരു കമാൻഡ് അടിസ്ഥാനമാക്കി ആപ്പ് ആണ്.
- യൂസർ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾക്ക് ഗ്രാഫിക്കൽ യൂസർ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ട്, അവയിൽ ബട്ടണുകൾ ക്ലിക്ക് ചെയ്യുകയും, ടെക്സ്റ്റ് ഇൻപുട്ട് ചെയ്യുകയും, विकल्पങ്ങൾ തിരഞ്ഞെടുക്കുകയും ചെയ്യാം.

### കോൺസോൾ ആപ്പുകളും UI ആപ്പുകളും പരിമിതമാണ്

ഇതിനെ കമാൻഡ് അടിസ്ഥാനമാക്കി ഒരു ആപ്പിനെ താരതമ്യം ചെയ്യുക, നിങ്ങൾ കമാൻഡ് ടൈപ്പ് ചെയ്യുന്നു:

- **ഇത് പരിമിതമാണ്**. നിങ്ങൾ ഏതു കമാൻഡും ടൈപ്പ് ചെയ്യാൻ കഴിയില്ല, ആപ്പ് പിന്തുണക്കുന്നത് മാത്രം.
- **ഭാഷാപ്രദിഷ്ഠിതമാണ്**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കാം, പക്ഷേ സാധാരണ ആപ്പ് ഒരു പ്രത്യേക ഭാഷക്കായി നിർമ്മിച്ചിരിക്കുന്നു, കൂടുതൽ ഭാഷകൾ ചേർക്കാൻ കഴിയും എങ്കിലും.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾക്ക് ലഭ്യമായ ഗുണങ്ങൾ

അതായത് ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിന്റെ വ്യത്യാസം എന്താണെന്ന്?

ഒരു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ നിങ്ങൾക്ക് കൂടുതൽ സ്വാതന്ത്ര്യമുണ്ട്, നിങ്ങൾ കമാൻഡുകളുടെ ഒരു സെറ്റിലോ ഒരു പ്രത്യേക ഇൻപുട്ട് ഭാഷയിലോ പരിമിതരായിട്ടില്ല. പകരം, നിങ്ങൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ഇടപഴകാം. മറ്റൊരു ഗുണം, നിങ്ങൾ ഇതിനകം വലിയ വിവരശേഖരം പരിശീലിപ്പിച്ചിരിക്കുന്ന ഡേറ്റാ ഉറവിടവുമായി ആശയവിനിമയം നടത്തുന്നു, സാധാരണ ആപ്പ് ഒരു ഡാറ്റാബേസിലുള്ള കാഴ്‌ച്ചകളിൽ മാത്രം പരിമിതപ്പെടുത്തപ്പെടാം.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ ഞാന് എന്തൊക്കെ നിർമ്മിക്കാമായി?

നിങ്ങൾക്ക് ബിൽഡ് ചെയ്യാൻ കഴിയുന്ന പല വസ്തുക്കളുണ്ട്. ഉദാഹരണങ്ങൾ:

- **ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനിയുടെയും അതിന്റെ ഉൽപ്പന്നങ്ങളുടെയും വിഷയങ്ങളെക്കുറിച്ചുള്ള ചോദ്യങ്ങൾക്ക് ഉത്തരം നൽകുന്ന ഒരു ചാറ്റ്ബോട്ട് നല്ല uyğunമാകാം.
- **ഹെല്പർ**. LLMകൾ ടെക്സ്റ്റ് സംഗ്രഹിക്കാൻ, ടെക്സ്റ്റിൽനിന്ന്洞察ങ്ങൾ നേടാൻ, റിസ്യൂമുകൾ പോലുള്ള ടെക്സ്റ്റ് ഉല്പാദിപ്പിക്കാൻ വലിയ സഹായം ചെയ്യുന്നു.
- **കോഡ് അസിസ്റ്റന്റ്**. നിങ്ങൾ ഉപയോഗിക്കുന്ന ഭാഷാ മോഡലിന്റെ അടിസ്ഥാനത്തിൽ, നിങ്ങൾക്ക് കോഡ് എഴുതാൻ സഹായിക്കുന്ന ഒരു കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉൽപ്പന്നവും ChatGPTയും കോഡ് എഴുതുന്നതിനായി സഹായിക്കാം.

## എങ്ങനെ ആരംഭിക്കാം?

LLM-ടൊടെ സംയോജിപ്പിക്കാൻ സഹായിക്കുന്ന രണ്ട് സാധാരണ സമീപനങ്ങൾ ഇങ്ങനെ:

- API ഉപയോഗിക്കുക. ഇവിടെ നിങ്ങൾ നിങ്ങളുടെ പ്രോംപ്റ്റിൽ വെബ് അഭ്യർത്ഥനകൾ നിർമ്മിച്ച് ഉൽപ്പന്നമായ ടെക്സ്റ്റ് തിരികെ ലഭിക്കും.
- ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API കോൾസ് ഉൾകൊള്ളിക്കുകയും ഉപയോഗിക്കാൻ എളുപ്പമാക്കുകയും ചെയ്യും.

## ലൈബ്രറികൾ/SDKകൾ

LLMകൾക്കായി കുറച്ചു പ്രശസ്തമായ ലൈബ്രറികളുണ്ട്, ഉദാഹരണങ്ങൾ:

- **openai**, ഈ ലൈബ്രറി മോഡലുമായി കണക്ട് ചെയ്യാനും പ്രോംപ്റ്റുകൾ നേരിട്ട് അയക്കാനും അത്യന്തം എളുപ്പമാക്കുന്നു.

അതിനാൽ ഒരു ഉയർന്ന തലത്തിലുള്ള ലൈബ്രറികളും ഉണ്ട്, ഉദാഹരണങ്ങൾ:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python പിന്തുണയുണ്ട്.
- **Semantic Kernel**. Semantic Kernel മൈക്രോസോഫ്റ്റിന്റെ ലൈബ്രറിയാണ്, C#, Python, Java ഭാഷകൾക്ക് പിന്തുണയുണ്ട്.

## openai ഉപയോഗിച്ച് ആദ്യ ആപ്പ്

നമുക്ക് ആദ്യ ആപ്പ് എങ്ങനെ നിർമ്മിക്കാമെന്ന്, ഏത് ലൈബ്രറികൾ ആവശ്യമാണെന്ന്, എത്ര ആവശ്യമാണ് തുടങ്ങിയവ കാണാം.

### openai ഇൻസ്റ്റാൾ ചെയ്യുക

  > [!NOTE] Codespaces-ല്‍ അല്ലെങ്കിൽ Devcontainer-ല്‍ ഈ നോട്ട്ബുക്ക് ഓടിച്ചാൽ ഈ ഘട്ടം വേണ്ട.


OpenAI അല്ലെങ്കിൽ Azure OpenAI-ഉം ചേർന്ന് പ്രവർത്തിക്കാൻ പല ലൈബ്രറികളും ഉണ്ട്. നിരവധി പ്രോഗ്രാമിങ് ഭാഷകളും ഉപയോഗിക്കാവുന്നതാണ്, C#, Python, JavaScript, Java എന്നിവ ഉൾപ്പെടെ.  
നാം `openai` Python ലൈബ്രറി തിരഞ്ഞെടുക്കുകയും അതിനായി `pip` ഉപയോഗിച്ച് ഇൻസ്റ്റാൾ ചെയ്യാനാണ് ഇതുവരെ കൂട്ടിച്ചേർത്തത്.

```bash
pip install openai
```

ഇത് Codespaces അല്ലെങ്കിൽ Dev Container-ൽ ഓടിക്കുന്നില്ലെങ്കിൽ, നിങ്ങളുടെ മെഷീനിൽ [Python](https://www.python.org/) ഇൻസ്റ്റാൾ ചെയ്തിരിക്കണം.

### ഒരു റിസോഴ്സ് സൃഷ്ടിക്കുക

നിങ്ങൾ ഇതുവരെ ചെയ്യാതെയുണ്ടെങ്കിൽ, താഴെപ്പറയുന്നവ ചെയ്യുക:

- Azure-യിൽ അക്കൗണ്ട് സൃഷ്ടിക്കുക <https://azure.microsoft.com/free/>.
- [Microsoft Foundry](https://ai.azure.com?WT.mc_id=academic-105485-koreyst)യിൽ Azure OpenAI റിസോഴ്സ് സൃഷ്ടിക്കുക. [റിസോഴ്സ് എങ്ങനെ സൃഷ്ടിക്കാമെന്ന്](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) ഈ ഗൈഡ് കാണുക.


### API കീയും എന്റ്പോയിന്റും കണ്ടെത്തുക

ഇപ്പോൾ നിങ്ങളുടെ `openai` ലൈബ്രറിയിൽ ഏത് API കീ ഉപയോഗിക്കണമെന്ന് बतിക്കണം. നിങ്ങളുടെ API കീ കണ്ടെത്താൻ, Azure OpenAI റിസോഴ്സിന്റെ "കീസ് ആൻഡ് എൻഡ്ബോയിന്റ്" വിഭാഗത്തിൽ പോയി "കീ 1" മൂല്യം കോപ്പി ചെയ്യുക.

  ![Azure പോർട്ടലിലുള്ള കീസും എൻഡ്ബോയിന്റും](https://learn.microsoft.com/azure/ai-foundry/openai/media/quickstarts/endpoint.png?WT.mc_id=academic-105485-koreyst)

ഇപ്പോൾ ഈ വിവരം കോപ്പി ചെയ്തു കഴിഞ്ഞതിനാൽ, ലൈബ്രറികൾ ഇതിനായി ഉപയോഗിക്കാനായി നിർദേശിക്കുന്നു.

> [!NOTE]
> നിങ്ങളുടെ API കീ കോഡിൽ നിന്ന് വേർപെടുത്തുന്നത് നല്ലതാണ്. ഇതിന് environment variable-കൾ ഉപയോഗിക്കാം.
> - `.env` ഫയലിൽ `AZURE_OPENAI_API_KEY` എന്ന environment variable-ലെ മൂല്യം നിങ്ങളുടെ API കീ ആയി സജ്ജമാക്കുക. ഈ കോഴ്സ് മുൻപ് പൂർത്തിയാക്കിയവർക്ക് ഇത് തയ്യാറായിരിക്കും.


### Azure കോൺഫിഗറേഷൻ സജ്ജമാക്കുക

നിങ്ങൾ Azure OpenAI ഉപയോഗിക്കുകയാണെങ്കിൽ, ഇതാ കോൺഫിഗറേഷൻ സജ്ജമാക്കാനുള്ള വിധം. Responses API Azure OpenAI **v1 endpoint**-ലാണ് സേവനം നൽകുന്നത്, അതുകൊണ്ട് `OpenAI` ക്ലയന്റിനെ `<your-endpoint>/openai/v1/`-നായി സൂചിപ്പിക്കുക:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']
```

മുകളിൽ ഞങ്ങൾ താഴെപ്പറയുന്നവ സജ്ജമാക്കുന്നു:

- `api_key`, ഇത് നിങ്ങളുടെ Azure പോർട്ടലിൽ നിന്നുള്ള API കീ ആണ്.
- `base_url`, നിങ്ങളുടെ Azure OpenAI എൻഡ്ബോയിന്റ് ആണ്, `/openai/v1/` appendedഉം. API കീ പക്കൽ എൻഡ്തിൽ തുടർന്നുള്ള `api_version` നൽകേണ്ടതില്ല.
- `deployment`, നിങ്ങൾ Foundry പോർട്ടലിൽ സൃഷ്ടിച്ച മോഡൽ ഡിപ്ലോയ്മെന്റ് നാമം.

> [!NOTE]
> `os.environ` എന്നത് environment variable-കൾ വായിക്കുന്ന ഒരു മാപ്പിംങാണ്. ഇത് `AZURE_OPENAI_API_KEY`, `AZURE_OPENAI_ENDPOINT` പോലുള്ള variable-കൾ വായിക്കാൻ ഉപയോഗിക്കാം.

## ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യുക

ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യാൻ `responses.create` മെത്തഡ് ഉപയോഗിക്കാം. ഉദാഹരണത്തോടെ കാണാം:

```python
prompt = "Complete the following: Once upon a time there was a"

```

മുകളിൽ കാഴ്ച വച്ച കോഡിൽ, നാം ഒരു റെസ്പോൺസ് ഔബ്ജക്ട് സൃഷ്ടിച്ച് ഉപയോഗിക്കാൻ ആഗ്രഹിക്കുന്ന മോഡലും പ്രോംപ്റ്റും നൽകുന്നു. പിന്നെ ജനിതകമായ ടെക്സ്റ്റ് പ്രിന്റ് ചെയ്യുന്നു.

### ചാറ്റ് റെസ്പോൻസ്

Responses API ഒറ്റ ടേൺ ടെക്സ്റ്റ് ജനറേഷനും മൾട്ടി ടേൺ ചാറ്റ്ബോട്ടിനും അനുയോജ്യമാണ് - `input` ലിസ്റ്റിൽ കൂടുതൽ മെസ്സേജുകൾ നൽകി സംഭാഷണം നിർമ്മിക്കുക:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],

  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

## വ്യത്യസ്ത കാര്യങ്ങൾക്കായി വ്യത്യസ്ത തരത്തിലുള്ള പ്രോംപ്റ്റുകൾ

ഇപ്പോൾ നിങ്ങൾക്ക് ഒരു പ്രോംപ്റ്റ് ഉപയോഗിച്ച് ടെക്സ്റ്റ് എങ്ങനെ ഉണ്ടാക്കാമെന്ന് നിങ്ങൾ കണ്ടതാണ്. നിങ്ങൾക്ക് ഒരു പ്രോഗ്രാം നിലവിലുണ്ടെന്നുവരുത്താനായിട്ടുണ്ട്, അത് നിങ്ങൾ മാറ്റിക്കൊണ്ട് വ്യത്യസ്ത തരത്തിലുള്ള ടെക്സ്റ്റ് ഉണ്ടാക്കാം.

പ്രോംപ്റ്റുകൾ വിവിധ ജോലികൾക്കായി ഉപയോഗിക്കാം. ഉദാഹരണത്തിന്:

- **ഒരു തരത്തിലുള്ള ടെക്സ്റ്റ് ജനനമാക്കുക**. ഉദാഹരണത്തിന്, ഒരു കവിത, ക്വിസിനുള്ള ചോദ്യങ്ങൾ എന്നിവ ഉണ്ടാക്കാം.
- **വിവരങ്ങൾ തിരയുക**. പ്രോംപ്റ്റുകൾ ഉപയോഗിച്ച് "വെബ് ഡവലപ്മെന്റിലെ CORS-ഉം എന്നർത്ഥം എന്ത്?" പോലുള്ള വിവരങ്ങൾ അന്വേഷിക്കാം.
- **കോഡ് ജനനം**. പ്രോംപ്റ്റുകളുടെ സഹായത്തോടെ ഇമെയിലുകൾ പരിശോധിക്കാൻ വേണ്ട റെഗുലർ എക്‌സ്‌പ്രഷൻ നിർമ്മിക്കുകയോ, ഒരു വെബ് ആപ്പ് പോലുള്ള മുഴുവൻ പ്രോഗ്രാം ഉണ്ടാക്കുകയോ ചെയ്യാം.

ഇപ്പോൾ നമുക്ക് ഒരുസന്നിവേശം കളിച്ചുകഴിഞ്ഞു, കാണിച്ച സന്നിവേശം അനുസരിച്ച് കോഡ് എഴുതാം. അതിനായി, താഴെപറഞ്ഞവ ചെയ്യുക:

1. നിലവിലുള്ള *app.py* ഫയൽ ഒരു ആരംഭ നിലവാരമായി ഉപയോഗിക്കുക
1. `prompt` വ്യത്യാസം കണ്ടെത്തുക, അതിന്റെ കോഡ് താഴെപറഞ്ഞതായിട്ട് മാറ്റുക:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

### Improve by adding filter and shopping list

    prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used, no {filter}"

- **Produce a shopping list**. We want to produce a shopping list, considering what we already have at home.

    For this functionality, we could either try to solve everything in one prompt or we could split it up into two prompts. Let's try the latter approach. Here we're suggesting adding an additional prompt, but for that to work, we need to add the result of the former prompt as context to the latter prompt. 

    Locate the part in the code that prints out the result from the first prompt and add the following code below:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # print response
    print("Shopping list:")
    print(response.output_text)
    ```

    Note the following:

    - We're constructing a new prompt by adding the result from the first prompt to the new prompt: 
    
        ```python
        new_prompt = f"{old_prompt_result} {prompt}"
        messages = [{"role": "user", "content": new_prompt}]
        ```

    - We make a new request, but also considering the number of tokens we asked for in the first prompt, so this time we say `max_output_tokens` is 1200. 

        ```python
        response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

   To change tokens used, you can use the `max_output_tokens` parameter. For example, if you want to use 100 tokens, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=100, store=False)
    ```

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs. The higher the temperature value the more random the output will be. Conversely the lower the temperature value the more predictable the output will be. Consider whether you want variation in your output or not.

   To alter the temperature, you can use the `temperature` parameter. For example, if you want to use a temperature of 0.5, you would do:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```
