# ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്ലിക്കേഷനുകൾ നിർമ്മിക്കുക

നിങ്ങൾ ഇതുവരെ ഈ പാഠ്യപദ്ധതിയിൽ കണ്ടിരിക്കുന്നത് പ്രോമ്പ്റ്റുകൾ പോലുള്ള മുഖ്യ ആശയങ്ങളും, hatta "prompt engineering" എന്ന ഒരു മുഴുവന്റെ ശാഖയും ഉണ്ട്. ChatGPT, Office 365, Microsoft Power Platform എന്നിവ പോലുള്ള പല ഉപകരണങ്ങളും നിങ്ങൾ പ്രോമ്പ്റ്റുകൾ ഉപയോഗിച്ച് ലക്ഷ്യം കൈവരിക്കുന്നതിന് സഹായിക്കുന്നു.

ഒരു ആപ്ലിക്കേഷനിൽ ഇത്തരത്തിലുള്ള അനുഭവം ചേർക്കാൻ, നിങ്ങൾക്ക് പ്രോമ്പ്റ്റുകൾ, ക്ലോസ്‌ഫിഷനുകൾ തുടങ്ങിയ ആശയങ്ങൾ മനസ്സിലാക്കണം, കൂടാതെ പ്രവർത്തിക്കാൻ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കണം. ഇത് തന്നെയാണ് ഈ അധ്യായത്തിൽ നിങ്ങൾ പഠിക്കേണ്ടത്.

## പരിചയം

ഈ അധ്യായത്തിൽ നിങ്ങൾ ചെയ്യും:

- openai ലൈബ്രറിയെക്കുറിച്ച് പ്രഭാഷണം ചെയ്യുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ പ്രോമ്പ്റ്റ്, താപനില, ടോക്കൺ എന്നിവ പോലുള്ള ആശയങ്ങൾ എങ്ങനെ ഉപയോഗിക്കാമെന്ന് മനസ്സിലാക്കുക.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠത്തിന് ശേഷം, നിങ്ങൾക്ക് കഴിയും:

- ഒരു ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കുക.
- openai ഉപയോഗിച്ച് ഒരു ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- കൂടുതൽ അല്ലെങ്കിൽ കുറവായ ടോക്കൺ ഉപയോഗിക്കാൻ നിങ്ങളുടെ ആപ്പ് ക്രമീകരിക്കുകയും, വ്യത്യസ്ത ഫലങ്ങൾക്കായി താപനില മാറ്റുകയും ചെയ്യുക.

## ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി ഒരു ആപ്പ് നിങ്ങൾ നിർമ്മിക്കുമ്പോൾ ഇതുപോലുള്ള ഒരു ഇന്റർഫേസ് ഉണ്ടാകും:

- കമാൻഡ് അടിസ്ഥാനമായതാണ്. കൺസോൾ ആപ്ലിക്കേഷനുകൾ എങ്ങനെ കമാൻഡ് ടൈപ്പ് ചെയ്ത് ഒരുടാസ്‌ക്ക് നടത്താം എന്നതാണ്. ഉദാഹരണത്തിന്, `git` ഒരു കമാൻഡ് അടിസ്ഥാനമുള്ള ആപ്പ് ആണ്.
- ഉപയോക്തൃ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾ ഗ്രാഫിക്കൽ ഉപയോക്തൃ ഇന്റർഫേസുകൾ(GUIs) ഉണ്ട്, ഇവയിൽ ബട്ടണുകൾ ക്ലിക്കു ചെയ്യുക, എഴുത്ത് നൽകുക, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കുക തുടങ്ങിയവയുണ്ട്.

### കൺസോൾനും UI ആപ്പുകളും പരിമിതമാണ്

നിങ്ങൾ കമാൻഡ് ടൈപ്പ് ചെയ്യുന്ന ഒരു കമാൻഡ് അടിസ്ഥാന ആപ്പുമായി താരതമ്യം ചെയ്യുക: 

- **പരിമിതമാണ്**. നിങ്ങൾ ഏതെങ്കിലും കമാൻഡ് ടൈപ്പ് ചെയ്യാൻ പോയാൽ, ആപ്പ് പിന്തുണയ്ക്കുന്നവ മാത്രം ആയിരിക്കും.
- **ഭാഷാ പ്രത്യേകമാണ്**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കാം, പക്ഷേ ഡീഫോൾട്ടായി ആപ്പ് ഒരു പ്രത്യേക ഭാഷയ്ക്ക് രൂപകൽപ്പന ചെയ്‌തിരിക്കും, കൂടുതൽ ഭാഷാ പിന്തുണ ചേർക്കാവുന്നതാണ്.

### ടെക്‌സ്‌റ് ജനറേഷൻ ആപ്പുകളുടെ ഗുണങ്ങൾ

അതിനാൽ ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പ് എങ്ങനെ വ്യത്യാസപ്പെട്ടിരിക്കുന്നു?

ടെക്‌സ്‌റ് ജനറേഷൻ ആപ്പിൽ നിങ്ങൾക്ക് കൂടുതൽ സ്വാതന്ത്ര്യം ഉണ്ട്, നിങ്ങളുടെ ഇൻപുട്ട് ഭാഷയ്ക്ക് പരിധിയില്ല, നിങ്ങൾ നാച്ചുറൽ ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ഇടപഴക്കം നടത്താം. കൂടാതെ, ഇത് ഒരു വലിയ വിവരശേഖരത്തിലുള്ള ഒരു ഡാറ്റാ സോഴ്സ് ഉപയോഗിക്കുന്നത് മൂലം, പരമ്പരാഗത ആപ്പുകൾക്ക് ലഭ്യമല്ലാത്ത ഒരു വിശാലമായ പരിശീലനത്തിനുള്ള ആക്‌സസ് നൽകുന്നു.

### ടെക്‌സ്‌റ്റ് ജനറേഷൻ ആപ്പിൽ എന്തെന്തു നിർമ്മിക്കാം?

നിർമ്മിക്കാൻ നിങ്ങൾക്ക് പലതും കഴിയും. ഉദാഹരണത്തിന്:

- **ഒരു ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനി, ഉൽപ്പന്നങ്ങൾ എന്നിവയെക്കുറിച്ചുള്ള ചോദ്യങ്ങൾക്ക് ഉത്തരം നൽകുന്ന ഒരു ചാറ്റ്ബോട്ട് ഒരു നല്ല തിരഞ്ഞെടുപ്പായിരിക്കും.
- **സഹായകനായി**. LLMകൾ സാരാംശം തയ്യാറാക്കൽ, റിപ്പോർട്ടുകൾ നൽകൽ പോലുള്ള പല കാര്യങ്ങളിലും മികച്ചതാണ്.
- **കോഡ് അസിസ്റ്റന്റ്**. നിങ്ങൾ തിരഞ്ഞെടുക്കുന്ന ഭാഷാമോഡലിന്റെ അടിസ്ഥാനത്തിൽ, നിങ്ങൾക്ക് ഒരു കോഡ് എഴുതാൻ സഹായിക്കുന്ന അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉൽപ്പന്നങ്ങളും ChatGPTയും കോഡ് എഴുതാൻ സഹായിക്കും.

## എങ്ങനെ ആരംഭിക്കാം?

ഒത്തുചേർത്ത് പ്രവർത്തിക്കാൻ LLMയുമായി ചേരാനുള്ള രണ്ട് സമാന മാർഗങ്ങൾ ഇതാണ്:

- API ഉപയോഗിക്കുക. ഇതിൽ നിങ്ങൾതീർ‍ഞ്ഞെടുക്കുന്ന പ്രോമ്പ്റ്റ് കൊണ്ട് വെബ് അഭ്യർത്ഥനകൾ അയച്ചും തിരിച്ചടിയിലുണ്ടാകുന്ന ടെക്‌സ്‍റ്റ് ലഭിക്കാം.
- ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API കോൾസ് എൻക്യാപ്സുലേറ്റ് ചെയ്ത് ഉപയോഗം എളുപ്പമാക്കുന്നു.

## ലൈബ്രറികൾ/SDKകൾ

LLMs ഉപയോഗിച്ച് പ്രവർത്തിക്കാൻ ചില പ്രശസ്തമായ ലൈബ്രറികൾ:

- **openai**, ഈ ലൈബ്രറി നിങ്ങളുടെ മോഡലുമായി ബന്ധപ്പെടാനും പ്രോമ്പ്റ്റുകൾ അയയ്ക്കാനും എളുപ്പമാക്കുന്നു.

മുകളിൽ പ്രവർത്തിക്കുന്ന മറ്റ് ലൈബ്രറികളും ഉണ്ട്:

- **Langchain**. Langchain പ്രശസ്തമാണ്, Python പിന്തുണയ്ക്കുന്നു.
- **Semantic Kernel**. Microsoft-ന്റെ Semantic Kernel C#, Python, Java ഭാഷകളെ പിന്തുണയ്ക്കുന്ന ലൈബ്രറിയാണ്.

## ChatCompletions ഉപയോഗിച്ച് ടെക്‌സ്‌റ്റ് ജനറേറ്റ് ചെയ്യുക

ടെക്‌സ്‌റ്റ് ജനറേറ്റ് ചെയ്യാനുള്ള മാർഗം `ChatCompletionsClient` ക്ലാസ് ആണ്.
`samples/python/azure_ai_inference/basic.py` എന്ന കോഡിൽ പ്രതികരണ ഭാഗത്ത്, യൂസർ റോളിലെ കോൺ‌റന്റ് പാരാമീറ്റർ താഴെപ്പറയുന്നതുപോലെ മാറ്റുക:

## വ്യത്യസ്ത കാര്യങ്ങൾക്ക് വ്യത്യസ്ത prompt-കൾ

ഇപ്പോൾ prompt ഉപയോഗിച്ച് എങ്ങനെ പാഠ്യം തയാറാക്കാമെന്ന് നിങ്ങൾക്ക് കണ്ടിട്ടുണ്ട്. ഒപ്പം വ്യത്യസ്ത തരത്തിലുള്ള പാഠ്യം ഉണ്ടാക്കാൻ നിങ്ങൾക്ക് മാറ്റം വരുത്താനും പിരഞ്ഞെടുക്കാനും സാധിക്കുന്ന ഒരു പ്രോഗ്രാം പ്രവർത്തനക്ഷമമാണ്.

Prompt-കൾ എല്ലാ തരം പ്രവർത്തനങ്ങൾക്കും ഉപയോഗിക്കാവുന്നത് ആണ്. ഉദാഹരണത്തിന്:

- **ഒരു തരം പാഠ്യം തയാറാക്കുക**. ഉദാഹരണത്തിന്, നിങ്ങൾക്ക് കവിത, ക്വിസ് ചോദ്യങ്ങൾ തുടങ്ങിയവ തയാറാക്കാൻ കഴിയും.
- **വിവരം അന്വേഷിക്കുക**. നിങ്ങൾക്ക് ഇങ്ങനെ ചോദിക്കുന്ന prompt-കൾ ഉപയോഗിച്ച് വിവരങ്ങൾ അന്വേഷിക്കാം; ഉദാഹരണത്തിന് 'വെബ് ഡെവലപ്പ്മെന്റിൽ CORS-ന്റെ അര്‍ത്ഥം എന്ത്?'.
- **കോഡ് സൃഷ്ടിക്കുക**. ഇമെയിൽ സാധുത പരിശോധിക്കാൻ ഉപയോഗിക്കുന്ന റഗുലർ എക്‌സ്പ്രഷൻ വികസിപ്പിക്കുന്നതിനോ, ഒരു മുഴുവൻ പ്രോഗ്രാം, ഉദാഹരണത്തിന് വെബ് ആപ്ലിക്കേഷൻ ഉണ്ടാക്കുന്നതിനോ prompt-കൾ ഉപയോഗിക്കാം.

1. ഒരു `prompt` വ്യത്യാസം സൃഷ്ടിച്ച് താഴെ കാണിച്ചിരിക്കുന്നതുപോലെ സാമ്പിൾ കോഡ് മാച് ചെയ്യുക:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

1. താഴെയുള്ള രീതിയിൽ നാം കോഡ് മാറ്റാം:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
