# ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകൾ നിർമ്മിക്കുക

ഇതുവരെ നിങ്ങൾ ഈ പാഠപുസ്തകത്തിലൂടെ നോക്കിയപ്പോൾ, പ്രോപ്റ്റുകൾ പോലുള്ള അടിസ്ഥാന ആശയങ്ങൾ ഉണ്ട് എന്നതും "പ്രോപ്റ്റ് എൻജിനീയറിംഗ്" എന്ന മുഴുവൻ ഒരു ശാസ്ത്രശാഖയും ഉണ്ട് എന്നതും നിങ്ങൾ കണ്ടിട്ടുണ്ട്. ChatGPT, Office 365, Microsoft Power Platform എന്നിവ പോലെയുള്ള പല ഉപകരണങ്ങളും, ഒന്നൊന്നിനെക്കുറിച്ച് എന്താമെങ്കിലും നേടാൻ നിങ്ങളെ പ്രോപ്റ്റുകൾ ഉപയോഗിച്ച് പിന്തുണയ്ക്കുന്നു.

ഒരു ആപ്പിൽ ഇത്തരം അനുഭവം ചേർക്കാനായി, പ്രോപ്റ്റുകൾ, പൂർത്തീകരണങ്ങൾ തുടങ്ങിയ ആശയങ്ങൾ മനസ്സിലാക്കേണ്ടതുണ്ട്, കൂടാതെ പ്രവർത്തിക്കാൻ ഒരു ലൈബ്രറി തിരഞ്ഞെടുക്കേണ്ടതുമാണ്. ഇന്നത്തെ അദ്ധ്യായത്തിൽ നിങ്ങൾ അതെല്ലാം പഠിക്കാം.

## പരിചയം

ഈ അദ്ധ്യായത്തിൽ, നിങ്ങൾ ചെയ്യും:

- openai ലൈബ്രറിയെയും അതിന്റെ അടിസ്ഥാന ആശയങ്ങളെയും കുറിച്ച് പഠിക്കുക.
- openai ഉപയോഗിച്ചു ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കുക.
- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ പ്രോപ്റ്റ്, താപനില, ടോകൺസ് മുതലായ ആശയങ്ങൾ എങ്ങനെ ഉപയോഗിക്കാമെന്ന് മനസ്സിലാക്കുക.

## പഠനലക്ഷ്യങ്ങൾ

ഈ പാഠം പൂർത്തിയാക്കുമ്പോൾ നിങ്ങൾ കഴിയും:

- ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണെന്ന് വിശദീകരിക്കാൻ.
- openai ഉപയോഗിച്ച് ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് നിർമ്മിക്കാൻ.
- കൂടുതൽ കുറച്ച് ടോകൺസ് ഉപയോഗിച്ച് ആപ്പ് കൺഫിഗർ ചെയ്യാനും, വിവിധ ഫലത്തിനായി താപനില മാറ്റാനും.

## ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് എന്താണ്?

സാധാരണയായി, നിങ്ങൾ ഒരു ആപ്പ് നിർമ്മിക്കുമ്പോൾ അതിന് ഏതമ്പോഴും ഒരു ഇന്റർഫേസ് വീതിയാണ് ഇതു പോലുള്ളത്:

- കമാൻഡ് അടിസ്ഥാനമാക്കിയുള്ളത്. കോൺസോൾ ആപ്പുകൾ സാധാരണയായി നിങ്ങൾ ഒരു കമാൻഡ് ടൈപ്പ് ചെയ്യുന്നു, അത് ഒരു പ്രവൃത്തി നിർവഹിക്കുന്നു. ഉദാഹരണത്തിന്, `git` ഒരു കമാൻഡ് അടിസ്ഥാനമാക്കിയുള്ള ആപ്പ് ആണ്.
- ഉപയോക്തൃ ഇന്റർഫേസ് (UI). ചില ആപ്പുകൾ ഗ്രാഫിക്കൽ യൂസർ ഇന്റർഫേസുകൾ (GUIs) ഉണ്ട്, അവിടെ നിങ്ങൾ ബട്ടണുകൾ ക്ലിക്ക് ചെയ്യുകയും, ടെക്സ്റ്റ് നൽകുകയും, ഓപ്ഷനുകൾ തിരഞ്ഞെടുക്കുകയും ചെയ്യാം.

### കോൺസോൾ ആപ്പുകളും UI ആപ്പുകളും സീമിതമാണ്

നിങ്ങളുടെ കമാൻഡ് മുറിച്ച് ചിന്തിക്കുക:

- **ഇത് സീമിതമാണ്**. നിങ്ങൾക്ക് ഏത് കമാൻഡ് വേണമെങ്കിലും ടൈപ്പ് ചെയ്യാൻ കഴിയില്ല, ആപ്പ് പിന്തുണയ്ക്കുന്നവ മാത്രമേ ഉപയോഗിക്കാറുള്ളൂ.
- **ഭാഷാ പ്രത്യേകമാണ്**. ചില ആപ്പുകൾ പല ഭാഷകളും പിന്തുണയ്ക്കുന്നു, എന്നാൽ സാധാരണ ആപ്പ് ഒരു പ്രത്യേക ഭാഷയ്‌ക്ക് മാത്രമേ നിർമ്മിച്ചുള്ളൂ, കൂടാതെ നിങ്ങൾക്കു കൂടുതൽ ഭാഷാ പിന്തുണ ചേർക്കാം.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പുകളുടെ പ്രയോജനങ്ങൾ

അതിനാൽ ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പ് മറ്റൊന്നായി എന്താണ?

ടെക്സ്റ്റ് генераേഷൻ ആപ്പിൽ, നിങ്ങൾക്ക് കൂടുതൽ സ്വാതന്ത്ര്യമുണ്ട്, നിങ്ങൾ കമാൻഡ് സെറ്റിനോ പ്രത്യേക ഇൻപുട്ട് ഭാഷയിലോ പരിമിതിക്കപ്പെടുന്നില്ല. പകരം, നിങ്ങൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് ആപ്പുമായി ആശയവിനിമയം നടത്താം. മറ്റൊരു പ്രയോജനം: നിങ്ങൾ ഇതിനകം വൻപഞ്ചായത്തിലൊരു ഡാറ്റാ സ്രോതസ്സുമായി ബന്ധപ്പെടുന്നു, ഇത് നേരത്തെ പരിശീലിപ്പിച്ച വിവരങ്ങളുടെ വലിയ സംസ്കാരം, സാധാരണ ആപ്പുകളിൽ ഡാറ്റാബേസിൽ വെച്ചിരിക്കുന്നതിൽ കുഴപ്പമുണ്ട്.

### ടെക്സ്റ്റ് ജനറേഷൻ ആപ്പിൽ എന്തെന്ത് നിർമ്മിക്കാമെന്ന് ഞാൻ എന്താണ് ചെയ്യേണ്ടത്?

നിരവധി പ്രവൃത്തികൾ നിങ്ങള്ക്ക് ചെയ്യാൻ കഴിയും. ഉദാഹരണങ്ങൾ:

- **ഒരു ചാറ്റ്ബോട്ട്**. നിങ്ങളുടെ കമ്പനി, അതിന്റെ ഉൽപ്പന്നങ്ങളിൽ സംബന്ധിച്ച ചോദ്യങ്ങൾക്കു മറുപടി നൽകുന്ന ഒരു ചാറ്റ്ബോട്ട് നല്ല മത്സരം ആകാം.
- **സഹായകൻ**. LLMകൾ ടെക്സ്റ്റ് സമര്സീകരിക്കാൻ, ടെക്സ്റ്റിൽ നിന്ന് വിവരങ്ങൾ കണ്ടെത്താൻ, റിസ്യൂമുകൾ പോലുള്ള ടെക്സ്റ്റ് ഉണ്ടാക്കാൻ മികച്ചവയാണ്.
- **കോഡ് സഹായിക്കാരൻ**. നിങ്ങൾ ഉപയോഗിക്കുന്ന ഭാഷാ മാതൃക വികസിപ്പിച്ച്, എങ്കിൽ നിങ്ങൾക്ക് കോഡ് എഴുതാൻ സഹായിക്കുന്ന ഒരു കോഡ് അസിസ്റ്റന്റ് നിർമ്മിക്കാം. ഉദാഹരണത്തിന്, GitHub Copilot പോലുള്ള ഉല്പന്നങ്ങളും ChatGPT-യും നിങ്ങളെ സഹായിക്കാൻ ഉപയോഗിക്കാം.

## ഞാൻ എങ്ങനെ ആരംഭിക്കാം?

സാധാരണയായി LLM-സ്-ഓടൊപ്പം സംയോജിപ്പിക്കാൻ രണ്ട് പതിവ് മാർഗ്ഗങ്ങൾ ഉണ്ട്:

- API ഉപയോഗിക്കുക. ഇവിടെ നിങ്ങൾ നിങ്ങളുടെ പ്രോപ്റ്റ് ഉപയോഗിച്ച് വെബ് അഭ്യർത്ഥനകൾ നിർമ്മിക്കുകയും, ഉണ്ടാക്കിയ ടെക്സ്റ്റ് തിരികെ ലഭിക്കുകയും ചെയ്യും.
- ഒരു ലൈബ്രറി ഉപയോഗിക്കുക. ലൈബ്രറികൾ API കോൾസ് മറുകടക്കുകയും ഉപയോഗിക്കാൻ എളുപ്പമാക്കുകയും ചെയ്യുന്നു.

## ലൈബ്രറികൾ/SDKകൾ

LLMs ഉപയോഗിക്കുന്നതിന് അറിയപ്പെടുന്ന കുറച്ച് ലൈബ്രറികൾ ഉണ്ട്:

- **openai**, ഈ ലൈബ്രറിയിലൂടെ നിങ്ങളുടെ മോഡലുമായി ബന്ധപ്പെടുകയും പ്രോപ്റ്റുകൾ അയയ്ക്കുകയും എളുപ്പമാണ്.

പിന്നീട് ഉയർന്ന തലത്തിലുള്ള ലൈബ്രറികൾ ഉണ്ട്:

- **Langchain**. Langchain നന്നായി അറിയപ്പെട്ടിരിക്കുന്നു, Python പിന്തുണയുള്ളത്.
- **Semantic Kernel**. Semantic Kernel മൈക്രോസോഫ്റ്റ് നിർമ്മിത ലൈബ്രറിയാണ്, C#, Python, Java തുടങ്ങിയ ഭാഷകളെ പിന്തുണയ്ക്കുന്നു.

### 2. VS Code (അഥവാ ഇഷ്ടപ്പെട്ട IDE)

നിങ്ങളുടെ Microsoft Foundry പ്രോജക്ടിൽ നിന്നുള്ള എല്ലാ വിവരങ്ങളുമുണ്ട്, നിങ്ങളുടെ ഇഷ്ടപ്പെട്ട IDE-യിൽ ഓടിക്കാൻ. ഈ ഉദാഹരണത്തിൽ VS Code കാണിക്കും

* നിങ്ങളുടെ ഫൗണ്ടറി പ്രോജക്ടിൽ ഒരു ചാറ്റ് മോഡൽ വിനിയോഗിക്കുക. ഈ പാഠത്തിലെ താപനില ഉദാഹരണങ്ങൾ **Llama** മോഡൽ പോലെയുള്ളവ ഉപയോഗിക്കുന്നു, ഉദാഹരണത്തിന് `Llama-3.3-70B-Instruct`, കാരണം `gpt-5-mini` പോലുള്ള രീസണിംഗ് മോഡലുകൾക്ക് `temperature` / `top_p` പിന്തുണയില്ല.
* പ്രോജക്ടിന്റെ **എൻഡ്പോയിന്റ്**യും **API കീ**യും Overview പേജിൽ നിന്ന് പകർത്തുക
* അവ സൂക്ഷിക്കാൻ പരിസ്ഥിതി ചുറ്റുമുള്ള വേരിയബിൾസ് സൃഷ്ടിക്കുക: `AZURE_INFERENCE_ENDPOINT` සහ `AZURE_INFERENCE_CREDENTIAL` - bash, powershell, windows കമാൻഡ് പ്രോംപ്റ്റ് സാമ്പിളുകൾ ലഭ്യമാണ്
* ഡിപ്പന്റൻസികൾ ഇൻസ്റ്റാൾ ചെയ്യുക: ```pip install azure-ai-inference```

## ChatCompletions ഉപയോഗിച്ച് ടെക്സ്റ്റ് ജനറേറ്റ് ചെയ്യുക

ടെക്സ്റ്റ് ഉണ്ടാക്കാനുള്ള മാർഗം `ChatCompletionsClient` ക്ലാസ് ഉപയോഗിക്കുകയാണ്.
`samples/python/azure_ai_inference/basic.py` ൽ, കോഡിന്റെ റിസ്പോൺസും പ്രദേശത്ത്, യൂസർ റോളിൽ ഉള്ള കോണ്ടന്റ് പാരാമീറ്റർ താഴെക്കാണുന്ന പോലെ അപ്ഡേറ്റ് ചെയ്യുക:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # ഐച്ഛിക പാരാമീറ്ററുകൾ
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

അപ്ഡേറ്റുചെയ്‌ത ഫയൽ ഓടിച്ച് ഫലം കാണുക


## വ്യത്യസ്ത കാര്യങ്ങൾക്ക് വ്യത്യസ്ത തരം പ്രാമ്പ്റ്റുകൾ

ഇപ്പോൾ നിങ്ങൾ പ്രാമ്പ്റ്റ് ഉപയോഗിച്ച് ടെക്സ്റ്റ് എങ്ങനെ ജനറേറ്റ് ചെയ്യാമെന്ന് കണ്ടു. നിങ്ങൾക്ക് ഇത് മാറ്റി വിവിധ തരം ടെക്സ്റ്റ് ഉത്പാദിപ്പിക്കാൻ കഴിയുന്ന ഒരു പ്രോഗ്രാം ഉണ്ടാക്കി പ്രവർത്തിപ്പിക്കാനും കഴിഞ്ഞിട്ടുണ്ട്.

പ്രാമ്പ്റ്റുകൾ വിവിധ ജോലികൾക്കായി ഉപയോഗിക്കാം. ഉദാഹരണത്തിന്:

- **ഒരു തരത്തിലുള്ള ടെക്സ്റ്റ് നിർമ്മിക്കുക**. ഉദാഹരണത്തിന്, കവിത, ക്വിസിനുള്ള ചോദ്യങ്ങൾ മുതലായവ നിർമ്മിക്കാൻ.
- **വിവരം തിരയുക**. ഉദാഹരണത്തിന്, 'വെബ് ഡെവലപ്പ്മെന്റിൽ CORS എപ്പോൾ എന്നാൽ എന്ത്?' എന്ന പോലുള്ള വിവരങ്ങൾ കണ്ടെത്താൻ.
- **കോഡ് നിർമ്മിക്കുക**. ഉദാഹരണത്തിന്, ഇമെയിൽ സാധൂകരിക്കാൻ ഉപയോഗിക്കുന്ന റെഗുലർ എക്സ്പ്രഷൻ എഴുതുക, അല്ലെങ്കിൽ ഒരു വെബ് ആപ്പ് പോലുള്ള മുഴുവൻ പ്രോഗ്രാം നിർമ്മിക്കുക.

## അഭ്യാസം - ഒരു റെസിപ്പി ജനറേറ്റർ നിർമ്മിക്കുക

ഇപ്പോൾ നമ്മൾ ഒരു സീനാരി വായിച്ചു കഴിഞ്ഞപ്പോൾ, പ്രദർശിപ്പിച്ച സീനാരിക്കു സമ്മതമാകുന്ന കോഡ് എഴുതാം. അതിനായി, ഈ ഘട്ടങ്ങൾ പിന്തുടരുക:

1. നിലവിലുള്ള ഫയൽ ആരംഭിക്കലായി ഉപയോഗിക്കുക
1. ഒരു `prompt` വാരിയബിൾ സൃഷ്ടിക്കുക, താഴെ കാണുന്നതുപോലെ സാമ്പിൾ കോഡ് മാറ്റുക:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

# temperature/top_p need a non-reasoning model (gpt-5 rejects them), so use a Llama model here.
model_name = os.environ.get("AZURE_INFERENCE_CHAT_MODEL", "Llama-3.3-70B-Instruct")

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

        ```output
        Feel free to omit the optional ingredients if you prefer or if you already have them on hand. Enjoy your delicious strawberry milkshake!
        ```
