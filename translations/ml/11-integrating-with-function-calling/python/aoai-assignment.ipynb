## ഫങ്ഷൻ കോൾകളെ മനസിലാക്കുക

ഈ പാഠത്തിനായി, ഞങ്ങൾ നമ്മുടെ ആധ്യാപന സ്റ്റാർട്ടപ്പിനായി ഒരു ഫീച്ചർ നിർമ്മിക്കാൻ ആഗ്രഹിക്കുന്നു, ഇത് ഉപയോക്താക്കൾക്ക് ഒരു ചാറ്റ്ബോട്ട് ഉപയോഗിച്ച് സാങ്കേതിക കോഴ്സുകൾ കണ്ടെത്താൻ അനുവദിക്കും. അവർക്ക് അവരുടെ നൈപുണ്യത്തിന്റെ തലശാരം, നിലവിലെ വേതനം, താൽപ്പര്യമുള്ള സാങ്കേതികവിദ്യ എന്നിവയെ ആശ്രയിച്ച് uygunമായ കോഴ്സുകൾ ശിപാർശ ചെയ്യുന്നു.

ഇതു പൂർത്തിയാക്കാൻ ഞങ്ങൾ താഴെ പറയുന്ന സംയോജനം ഉപയോഗിക്കാം: 
 - ഉപയോക്താവിന് ഒരു ചാറ്റ് അനുഭവം സൃഷ്ടിക്കാൻ `Azure Open AI`
 - ഉപയോക്താവിന്റെ അഭ്യർത്ഥനയുടെ അടിസ്ഥാനത്തിൽ കോഴ്സുകൾ കണ്ടെത്താൻ സഹായിക്കാൻ `Microsoft Learn Catalog API`
 - ഉപയോക്താവിന്റെ ചോദ്യം എടുത്ത് ഫങ്ഷനിലേക്ക് അയച്ചു API അഭ്യർത്ഥന നടത്താൻ `Function Calling`.

ആരംഭിക്കാൻ, ആദ്യം നമുക്ക് ഫങ്ഷൻ കോൾ ഉപയോഗിക്കാൻ കാരണം നോക്കാം: 


### ഫങ്ക്ഷൻ കോളിംഗ് എന്തിന് 

നിങ്ങൾ ഈ കോഴ്സിൽ മറ്റെന്തെങ്കിലും പാഠം പൂർത്തിയാക്കിയിട്ടുണ്ടെങ്കിൽ, വലിയ ഭാഷാ മോഡലുകൾ (LLMs) ഉപയോഗിക്കുന്നതിന്റെ കരുത്ത് നിങ്ങൾക്ക് മനസ്സിലാകും. അതുപോലെ, അവയുടെ ചില പരിമിതികൾ കാണാനും കഴിയുമെന്ന് പ്രതീക്ഷിക്കുന്നു. 

ഫങ്ക്ഷൻ കോളിംഗ് ആഷ്യൂർ ഓപ്പൺ AI സർവീസിന്റെ ഒരു സവിശേഷതയാണ്, താഴെ പറയുന്ന പരിമിതികൾ മറികടക്കാൻ: 
1) കൺസിസ്റ്റന്റ് പ്രതികരണ ഫോർമാറ്റ് 
2) ചാറ്റ് സന്ദർഭത്തിൽ ഒരു അപ്ലിക്കേഷനിലെ മറ്റു സ്രോതസ്സുകളില്‍ നിന്നുള്ള ഡാറ്റ ഉപയോഗിക്കാൻ സാദ്ധ്യത 

ഫങ്ക്ഷൻ കോളിംഗ് മുൻപ്, LLM-കളുടെ പ്രതികരണങ്ങൾ ഘടിതമല്ലാതെയും അസംഘടിതവുമായിരുന്ന. ഡിസൈനർമാർ ഓരോ പ്രതികരണ വ്യത്യാസവും കൈകാര്യം ചെയ്യാൻ സങ്കീർണമായ സാധുതാ കോഡ് എഴുതേണ്ടിവന്നു. 

"സ്റ്റോക്ക്ഹോംലിൽ ഇപ്പോൾ ഉള്ള കാലാവസ്ഥ എന്താണ്?" എന്നെത്തന്നെ ഉത്തരം ഉപഭോക്താക്കൾക്ക് ലഭ്യമാകാൻ കഴിയുന്നില്ലായിരുന്നു. കാരണം മോഡലുകൾ പഠിപ്പിച്ച ഡാറ്റാ സമയത്തിനു പരിമിതമായിരുന്നു. 

താഴെ കാണുന്ന ഉദാഹരണം ഈ പ്രശ്നം വിശദീകരിക്കുന്നു: 

വിദ്യാർത്ഥികളുടെ ഡേറ്റാബേസ് സൃഷ്ടിച്ച് അവര്ക്ക് ശരിയായ കോഴ്സ് നിർദ്ദേശിക്കാൻ ആഗ്രഹിക്കുന്നുവെന്ന് വിചാരിക്കാം. താഴെ രണ്ട് വിദ്യാർത്ഥികളുടെ വിവരങ്ങൾ നൽകിയിട്ടുണ്ട്, അതിൽ നിന്ന് ലഭിക്കുന്ന ഡാറ്റ വളരെ സമാനമാണ്.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ഡാറ്റ പ്രോസസ് ചെയ്യാൻ ഒരു LLM-ക്ക് അത് അയയ്ക്കാൻ ഞങ്ങൾ ആഗ്രഹിക്കുന്നു. പിന്നീട് ഇത് ഞങ്ങളുടെ ആപ്ലിക്കേഷനിൽ API-ക്ക് അയയ്ക്കാനോ ഒരു ഡാറ്റാബേസിൽ സംഭരിക്കാനോ ഉപയോഗിക്കാവുന്നതാണ്.

നമുക്ക് രണ്ട് സമാനമായ പ്രോംപ്റ്റുകൾ ഉണ്ടാക്കാം, അവയിൽ ഞങ്ങൾ LLM-ന് ഏത് വിവരത്തിൽ താൽപര്യമുള്ളുവെന്നു വ്യക്തമാക്കുന്നു:


ഞങ്ങള്‍ ഇത് ഒരു LLM-ല്‍ അയച്ച് നമ്മുടെ ഉല്‍പന്നത്തിന് പ്രധാനപ്പെട്ട ഭാഗങ്ങള്‍ പാഴ്‌സ് ചെയ്യാന്‍ ആഗ്രഹിക്കുന്നു. അതിനാല്‍ LLM നു നിര്‍ദ്ദേശം നല്‍കാന്‍ രണ്ട് സമാന പ്രമ്പ്റ്റുകള്‍ ഞങ്ങള്‍ രൂപകല്‍പ്പന ചെയ്യാം: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ഈ രണ്ട് പ്രാമ്പ്റ്റുകളും സൃഷ്ടിച്ചതിന് ശേഷം, നാം അവ `client.responses.create` ഉപയോഗിച്ച് LLM-ലേക്ക് അയയ്ക്കും. നാം പ്രാമ്പ്റ്റ് `input` വേരിയബിൾ-ലിൽ സംഭരിക്കുകയും റോളായി `user` നൽകുകയും ചെയ്യും. ഇത് ഒരു ഉപയോക്താവ് ചാറ്റ്ബോട്ട്-ലേക്ക് സന്ദേശം എഴുതുന്നതായി അനുകരിക്കുന്നതിനാണ്. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ഇപ്പാഴാണ് നമുക്ക് ഇരുഭാഗവും LLM-ലേക്ക് അയച്ച് നാം നല്‍കുന്ന പ്രതികരണം പരിശോധിക്കാനാകുന്നത്. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```

ഫങ്ഷണൽ കോൾ ചെയ്യലിലൂടെ നാം ഉറപ്പ് വരുത്താം പിന്‍വാങ്ങുന്ന ഡാറ്റ രൂപത്തിലിരിക്കുമെന്നും. ഫങ്ഷൻ കോൾ ചെയ്യുമ്പോൾ, LLM യഥാർത്ഥത്തിൽ യാതൊരു ഫങ്ഷനുകളെയും കോൾ ചെയ്യുകയോ ചലിപ്പിക്കുകയോ ചെയ്യില്ല. പകരം, LLM അതിന്റെ പ്രതികരണങ്ങൾക്കും പിന്തുടരാൻ ഒരു ഘടന സൃഷ്ടിക്കുന്നു. പിന്നീട് ആ ഘടനാപരമായ പ്രതികരണങ്ങൾ അടിസ്ഥാനമാക്കി നാം നമ്മുടെ ആപ്ലിക്കേഷനുകളിൽ ഏത് ഫങ്ഷൻ ഓടിക്കണം എന്ന് അറിയും.

ഉപയോക്താക്കൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് വിവരങ്ങൾ കണ്ടെത്തുന്നു, അവ ഫോർമാറ്റ് ചെയ്ത ക്വറി അല്ലെങ്കിൽ API അഭ്യർത്ഥനയായി മാറുന്നു. ഒരു ഉദാഹരണം, "അവസാന അസൈൻമെന്റ് പൂർത്തിയാക്കിയ വിദ്യാർത്ഥികൾ ആരൊക്കെയാണ്" എന്ന് ചോദിക്കുന്ന ഒരു അധ്യാപകനാണ്, ഇത് `get_completed(student_name: string, assignment: int, current_status: string)` എന്ന ഫംഗ്ഷൻ കോളാക്കി ചുരുക്കിയേക്കാം.


**ഘടനമാർന്ന ഡാറ്റ സൃഷ്ടിക്കൽ**
ഉപയോക്താക്കൾ ഒരു ബ്‌ലോക്കായ ടെക്സ്റ്റ് അല്ലെങ്കിൽ CSV എടുത്ത് LLM ഉപയോഗിച്ച് അതിൽ നിന്നുള്ള പ്രധാന വിവരങ്ങൾ എടുക്കാം. ഉദാഹരണത്തിന്, ഒരു വിദ്യാർത്ഥി സമാധാന കരാറുകൾ സംബന്ധിച്ച വിക്കിപീഡിയ ലേഖനം എടുക്കുകയും, AI ഫ്ലാഷ് കാർഡുകൾ സൃഷ്ടിക്കാനായി മാറ്റുകയും ചെയ്യാം. ഇത് `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` എന്ന ഫംഗ്ഷൻ ഉപയോഗിച്ച് സാധിക്കും.


## 2. നിങ്ങളുടെ ആദ്യ ഫംഗ്‌ഷൻ കോൾ സൃഷ്‌ടിക്കൽ 

ഒരു ഫംഗ്‌ഷൻ കോൾ സൃഷ്‌ടിക്കുന്ന പ്രക്രിയയിൽ 3 പ്രധാന ഘട്ടങ്ങൾ ഉണ്ട്: 
1. നിങ്ങളുടെ ഫംഗ്‌ഷനുകളുടെ ലിസ്റ്റും ഒരു ഉപയോക്തൃ സന്ദേശവും ഉപയോഗിച്ച് ചാറ്റ് സ്പഷ്ടീകരണങ്ങൾ API-നെ വിളിക്കുന്നത് 
2. ഒരു പ്രവർത്തനം ചെയ്യുന്നതിനായി മോഡലിന്റെ പ്രതികരണം വായിക്കുക, അതായത് ഒരു ഫംഗ്‌ഷൻ അല്ലെങ്കിൽ API കോൾ നടപ്പിലാക്കുക 
3. ഉപയോക്താവിന് മറുപടി സൃഷ്‌ടിക്കാൻ ആ വിവരങ്ങൾ ഉപയോഗിക്കുന്നതിനായി നിങ്ങളുടെ ഫംഗ്‌ഷനിൽ നിന്ന് ലഭിച്ച പ്രതികരണത്തോടെ ചാറ്റ് സ്പഷ്ടീകരണങ്ങൾ API-നെ വീണ്ടും വിളിക്കുക.

### ഫംഗ്ഷന്‍ കോളിന്റെ ഘടകങ്ങള്‍ 

#### ഉപയോക്തൃ ഇന്പുട്ട് 

ആദ്യ പടി ഒരു ഉപയോക്തൃ സന്ദേശം സൃഷ്ടിക്കുകയാണ്. ഇത് ഒരു ടെക്സ്റ്റ് ഇന്‍പുട്ടിന്റെ മൂല്യം സ്വീകരിക്കുന്നതോടെ ഡൈനാമിക് ആയിരിക്കും അല്ലെങ്കില്‍ നിങ്ങള്‍ക്ക് ഇവിടെ മൂല്യം അനുവദിക്കാം. ഇത് Chat Completions API ഉപയോഗിച്ച് ആദ്യമായി പ്രവര്‍ത്തിക്കുമ്പോള്‍, സന്ദേശത്തിന്റെ `role`നും `content`ഉം നിശ്ചയിക്കണം. 

`role` അപരിചിതമായാല്‍ `system` (നിയമങ്ങള്‍ സൃഷ്ടിക്കുന്നത്), `assistant` (മോഡല്‍) അല്ലെങ്കിൽ `user` (അവസാന ഉപയോക്താവ്) ആകാം. ഫംഗ്ഷന്‍ കോളിംഗിന്, നമ്മള്‍ ഇതിനെ `user` എന്ന് നിശ്ചയിച്ച് ഒരു ഉദാഹരണ ചോദ്യവും നല്‍കാം. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ഫംഗ്ഷനുകൾ സൃഷ്ടിക്കുന്നു. 

അടുത്തതായി ഒരു ഫംഗ്ഷനും ആ ഫംഗ്ഷന്റെ പാരാമീറ്ററുകളും നന്ദ് നിർവചിക്കും. ഇവിടെ `search_courses` എന്നൊരു ഫംഗ്ഷനേ മാത്രം ഉപയോഗിക്കും, പക്ഷേ നിങ്ങൾക്ക് പല ഫംഗ്ഷനുകളും സൃഷ്ടിക്കാൻ കഴിയും.

**പ്രധാനമാണ്** : ഫംഗ്ഷനുകൾ LLM-നു ലഭിക്കുന്ന സിസ്റ്റം സന്ദേശത്തിലായും അവർക്ക് ലഭ്യമായ ടോക്കൺ മാതൃകയിലായും ഉൾപ്പെടുത്തിയിരിക്കും. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**നിർവചനങ്ങൾ** 

`name` - വിളിക്കാനായി ആവശ്യമായ ഫംഗ്ഷന്റെ പേര്.

`description` - ഫംഗ്ഷൻ എങ്ങനെ പ്രവർത്തിക്കുന്നു എന്നതിന്റെ വിവരണം. ഇവിടെ വ്യക്തവും വ്യക്തമുമായിരിക്കണം.

`parameters` - മോഡൽ അതിന്റെ പ്രതികരണത്തിൽ ഉണ്ടാക്കാൻ നിങ്ങൾ ആഗ്രഹിക്കുന്ന മൂല്യങ്ങൾക്കും ഫോർമാറ്റിനും ഒരു പട്ടിക.


`type` - പ്രോപ്പർട്ടികൾ സൂക്ഷിക്കപ്പെടുന്ന ഡാറ്റാ ടൈപ്പ്.

`properties` - മോഡൽ അതിന്റെ പ്രതികരണത്തിൽ ഉപയോഗിക്കുന്ന പ്രത്യേക മൂല്യങ്ങളുടെ പട്ടിക.


`name` - മോഡൽ ഫോർമാറ്റുചെയ്ത പ്രതികരണത്തിൽ ഉപയോഗിക്കുന്ന പ്രോപ്പർട്ടിയുടെ പേര്.

`type` - ഈ പ്രോപ്പർട്ടിയുടെ ഡാറ്റാ ടൈപ്പ്.

`description` - പ്രത്യേക പ്രോപ്പർട്ടിക്ക് ഉള്ള വിവരണം.


**ഐച്ഛികം**

`required` - ഫംഗ്ഷൻ കോൾ പൂർത്തിയാക്കാൻ ആവശ്യമായ പ്രോപ്പർട്ടി.


### ഫങ്‌ഷൻ കോളുചെയ്യൽ  
ഒരു ഫങ്‌ഷൻ നിർവ്വചിച്ചതിന് ശേഷം, ഇപ്പോൾ അത് Chat Completion API-യുടെ കോളിൽ ഉൾപ്പെടുത്തേണ്ടതുണ്ട്. ഇത് അഭ്യർത്ഥനയിൽ `functions` ചേർത്തുകൊണ്ട് ചെയ്യാം. ഈ കേസിൽ `functions=functions` ആണ്.  

`function_call`-നെ `auto` ആയി സജ്ജീകരിക്കുന്ന ഓപ്ഷനും ഉണ്ട്. ഇത്ന്റെ അർത്ഥം, ഉപയോക്തൃ സന്ദേശത്തിന്റെ അടിസ്ഥാനത്തിൽ ഏത് ഫങ്‌ഷൻ കോളുചെയ്യണമെന്ന് LLM ആണ് തീരുമാനിക്കുന്നത്, നാം തന്നെ നിയമിക്കുന്നതിന് പകരം.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

## 3.ഒരു ആപ്ലിക്കേഷനിൽ ഫങ്ഷൻ കോൾസ് സംയോജിപ്പിക്കൽ.


LLM-ൽ നിന്നുള്ള ഫോർമാറ്റ് ചെയ്ത പ്രതികരണം പരീക്ഷിച്ചതിന് ശേഷം, ഇനി നാം ഈ കോൾ ഒരു ആപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കാം.

### പ്രവാഹം നിയന്ത്രിക്കൽ

ഇത് നമ്മുടെ ആപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കാൻ, നമുക്ക് താഴെപ്പറയുന്ന ഘട്ടങ്ങൾ സ്വീകരിക്കാം:

ആദ്യം, Open AI സേവനങ്ങളെ കോൾ ചെയ്ത്, സന്ദേശം `response_message` എന്ന വേരിയബിളിൽ സംഭരിക്കാം.


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ഇപ്പോൾ നാം കോഴ്സുകളുടെ പട്ടിക प्राप्तിക്കാൻ Microsoft Learn API വിളിക്കാൻ പോകുന്ന ഫങ്ഷൻ നിർവചിക്കാം: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

മികച്ച ഒരു രീതിയായി, മോഡൽ ഒരു ഫംഗ്ഷൻ വിളിക്കണമെന്ന് ആഗ്രഹിക്കുന്നുണ്ടോ എന്ന് നമുക്ക് ഇനി നോക്കാം. അതിനുശേഷം, ലഭ്യമായ ഫംഗ്ഷനുകളിൽ ഒന്നിനെ സൃഷ്ടിച്ച് ആ ഫംഗ്ഷൻ വിളിക്കപ്പെട്ടതുമായി പൊരുത്തപ്പെടുത്താം. 
തുടർന്ന്, ആ ഫംഗ്ഷന്റെ аргള്ന്മെന്റുകൾ എടുക്കുകയും അവയെ LLM-ൽ നിന്നുള്ള аргള്ന്മെന്റുകളുമായി മാപ്പ് ചെയ്യുകയും ചെയ്‌തുകൊള്ളാം.

അവസാനം, ഫംഗ്ഷൻ കോൾ സന്ദേശവും `search_courses` സന്ദേശം വഴിയായി തിരികെ ലഭിച്ച മൂല്യങ്ങളും നമുക്ക് ചേർക്കാം. ഇത് LLM-ന് ഉപയോക്താവിന്റെ പ്രകൃതഭാഷയിലുള്ള പ്രതികരണം നൽകാൻ ആവശ്യമായ എല്ലാ വിവരങ്ങളും നൽകുന്നു.
ഉപഭോക്താവിനോട് സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് പ്രതികരിക്കാൻ.


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ഇപ്പോൾ આપણે പുതുക്കിയ സന്ദേശം LLM-ക്ക് അയയ്ക്കാം, അതു കൊണ്ട് നാം API JSON ഫോർമാറ്റിലുള്ള മറുപടി അറിയാതെ ഒരു സ്വാഭാവിക ഭാഷാ പ്രതികരണം സ്വീകരിക്കാൻ കഴിയും.
