ഇതു പൂര്‍ത്തിയാക്കാന്‍, ഞങ്ങള്‍ താഴെ ചേര്‍ത്തുകളയുന്നവ ഉപയോഗിക്കും: 
 - ഉപയോക്താവിന് ചാറ്റ് അനുഭവം സൃഷ്‌ടിക്കാന്‍ `Azure Open AI`

 - ഉപയോക്താവിന്റെ ചോദ്യം Function Calling വഴി സ്വീകരിച്ച് API അഭ്യര്‍ത്ഥന നടത്താന്‍ ഒരോക фങ്ഷന്

തുടങ്ങാനായി, Function Calling ആദ്യമേ എന്തുകൊണ്ടാണ് വേണ്ടെന്ന് നോക്കാം: 


### ഫങ്ഷൻ കോളിംഗിന്റെ കാര്യം  

നിങ്ങൾ ഈ കോഴ്സ്中的 മറ്റേതെങ്കിലും പാഠം പൂർത്തിയാക്കിയിട്ടുണ്ടെങ്കിൽ, വലിയ ഭാഷാ മാതൃകകൾ (LLMs) ഉപയോഗിക്കുന്നതിന്റെ ശക്തി നിങ്ങൾക്കറിയാമാകും. അതേസമയം, അതിന്റെ ചില പരിമിതികളും നിങ്ങൾക്ക് കാണാമെന്നാണ് പ്രതീക്ഷ.  

ഫങ്ഷൻ കോളിംഗ് ആണ് Azure Open AI സർവീസിലെ ഒരു സൗകര്യം, താഴെപ്പറയുന്ന പരിമിതികളെ അതിക്രമിക്കാൻ:  
1) സ്ഥിരതയുള്ള പ്രതികരണ രൂപരേഖ  
2) ചാറ്റ് പ്രാസംഗികതയിൽ ആപ്പ് 的 മറ്റു സ്രോതസ്സുകളിൽ നിന്നുള്ള ഡാറ്റ ഉപയോഗിക്കാൻ കഴിവ്  

ഫങ്ഷൻ കോളിംഗ് വരുംമുമ്പ്, LLM-ലിൽ നിന്നുള്ള പ്രതികരണങ്ങൾ അസംഘടിതവും അനിശ്ചിതവുമായിരുന്നു. ഡെവലപ്പർമാർ ഓരോ വ്യത്യാസവും കൈകാര്യം ചെയ്യാൻ സങ്കീർണമായ പരിശോധന കോഡ് എഴുതേണ്ടിവന്നു.  

ഉപയോക്താക്കൾക്ക് "സ്റ്റോക്ക്‌ഹോംയിലെ നിലവിലെ കാലാവസ്ഥ എന്താണ്?" എന്ന പോലുള്ള ഉത്തരങ്ങൾ കിട്ടാൻ കഴിയില്ലായിരുന്നു. കാരണം, മോഡലുകൾ പരിശീലിപ്പിച്ച ഡാറ്റയുടെ സമയമായി പരിമിതമായിരുന്നു.  

ഈ പ്രശ്നം കാണിക്കുന്ന ഉദാഹരണം താഴെ നോക്കാം:  

വിദ്യാർത്ഥി ഡാറ്റയുടെ ഒരു ഡാറ്റാബേസ് സൃഷ്ടിച്ച് ഏത് കോഴ്‌സ് അവർക്ക് അനുയോജ്യമാണ് എന്ന് നിർദ്ദേശിക്കാനുള്ള ഉദ്ദേശ്യമാണ്. താഴെ കാണുന്ന രണ്ട് വിദ്യാർത്ഥികളുടെ വിവരണങ്ങൾ സൂക്ഷ്മമായി സമാനമാണ്.  


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ഡാറ്റ പാഴ്സുചെയ്യാൻ ഇത് ഒരു LLM-ന് അയക്കാൻ ഞങ്ങൾ ആഗ്രഹിക്കുന്നു. പിന്നീട് ഇത് ഞങ്ങളുടെ ആപ്ലിക്കേഷനിൽ API-യിലേക്കോ ഡേറ്റാബേസിലേക്കോ അയക്കാൻ ഉപയോഗിക്കാം. 

നമുക്ക് LLM-ന് ഞങ്ങൾ താൽപര്യമുള്ള വിവരങ്ങളെക്കുറിച്ച് നിർദ്ദേശിക്കാനുള്ള രണ്ട് സമാനമായ പ്രോംപ്റ്റുകൾ സൃഷ്ടിക്കാം: 


നമുക്ക് ഇത് ഒരു LLM-ക്ക് അയച്ചു, നമ്മുടെ ഉൽപ്പന്നത്തിനായി പ്രധാനമായ ഭാഗങ്ങൾ പാർസ് ചെയ്യിക്കാൻ ആഗ്രഹിക്കുന്നു. അതിനാൽ LLMയെ നിർദ്ദേശിക്കാൻ രണ്ട് ഒരുപോലെയുള്ള പ്രോംപ്റ്റുകൾ സൃഷ്ടിക്കാം: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ഈ രണ്ട് പ്രോംപ്റ്റുകളും സൃഷ്ടിച്ചശേഷം, ഞങ്ങൾ അവ `client.responses.create` ഉപയോഗിച്ച് LLM-ന് അയക്കുമെന്ന്.  പ്രോംപ്റ്റ് `input` വേരിയബിളിൽ സൂക്ഷിച്ച് റോളിന് `user` എന്ന് അസൈൻ ചെയ്യുന്നു. ഇത് ഒരു ഉപഭോക്താവ് ചാറ്റ്ബോട്ടിലേക്ക് മെസ്സേജ് എഴുതുന്നതിനെ അപര്യാപ്തരാക്കാൻ ആണ്.



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

# The OpenAI client points at the Azure OpenAI (Microsoft Foundry) /openai/v1/ endpoint
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment=os.environ['AZURE_OPENAI_DEPLOYMENT']
```

ഇപ്പോൾ നാം ഇരട്ട അഭ്യർത്ഥനകളും LLM-ലേക്ക് അയച്ച് നമുക്ക് ലഭിക്കുന്ന പ്രതികരണം പരിശോധിക്കാം. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```

ഉപയോക്താക്കൾ സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് വിവരങ്ങൾ കണ്ടെത്താൻ കഴിയും, അത് ഒരു ഫോർമാറ്റ് ചെയ്ത ക്വറി അല്ലെങ്കിൽ API അഭ്യർത്ഥനയാക്കി പരിവർത്തനം ചെയ്യപ്പെടുന്നു. ഉദാഹരണമായി, ഒരു അധ്യാപകൻ "അവസാന അസൈന്മെന്റ് പൂർത്തിയാക്കിയ വിദ്യാർത്ഥികൾ ആരെല്ലാ?" എന്നു ചോദിക്കുന്നത്, `get_completed(student_name: string, assignment: int, current_status: string)` എന്ന ഫംഗ്ഷൻ വിളിക്കാനും കഴിയും  


**സംരചിത ഡാറ്റ സൃഷ്ടിക്കൽ**  
ഉപയോക്താക്കൾ ഒരു ടെക്സ്റ്റ് ബ്ലോക്ക് അല്ലെങ്കിൽ CSV എടുത്ത് അതിൽ നിന്നുള്ള പ്രധാന വിവരങ്ങൾ LLM ഉപയോഗിച്ച് എക്സ്ട്രാക്ട് ചെയ്യാം. ഉദാഹരണമായി, ഒരു വിദ്യാർത്ഥി സമാധാന കരാറുകളെപ്പറ്റി ഉള്ള വിക്കിപീഡിയ ലേഖനം AI ഫ്‌ളാഷ് കാർഡുകൾ രൂപപ്പെടുത്താൻ മാറ്റാം. ഇത് `get_important_facts(agreement_name: string, date_signed: string, parties_involved: list)` എന്ന ഫംഗ്ഷൻ ഉപയോഗിച്ചുള്ളൊരു ഉദാഹരണമാണ്  


## 2. നിങ്ങളുടെ ആദ്യ ഫംഗ്ഷൻ കോളിന്റെ സൃഷ്ടി

ഫംഗ്ഷൻ കോൾ സൃഷ്ടിക്കുന്ന പ്രക്രിയയ്ക്ക് 3 പ്രധാന ഘട്ടങ്ങൾ ഉണ്ട്:

#### ഉപയോക്തൃ ഇൻപുട്ട്

ആദ്യം ചെയ്യേണ്ടത് ഉപയോക്തൃ സന്ദേശം സൃഷ്ടിക്കുക എന്നതാണ്. ഇത് ഒരു ടെക്സ്റ്റ് ഇൻപുട്ടിന്റെ മൂല്യം എടുത്ത് ഡൈനാമിക്കായി നിയോഗിക്കാനോ അല്ലെങ്കിൽ നിങ്ങൾക്ക് ഇവിടെ ഒരു മൂല്യം നിയോഗിക്കാനോ സാധിക്കും. ഇത് Chat Completions API ഉപയോഗിച്ച് നിങ്ങൾ ആദ്യമായി ജോലി ചെയ്യുമ്പോൾ, സന്ദേശത്തിന്റെ `role` ഉം `content` ഉം നിർവ്വചിക്കേണ്ടതാണ്.

`role` എന്നത് `system` (നിയമങ്ങൾ സൃഷ്ടിക്കുന്നത്), `assistant` (മോഡൽ) അല്ലെങ്കിൽ `user` (അവസാന ഉപയോക്താവ്) ആയിരിക്കാം. ഫംഗ്ഷൻ കോൾ ചെയ്യുമ്പോൾ, ഇത് `user` ആയി നിയോഗിക്കുകയും ഒരു ഉദാഹരണ ചോദ്യമുണ്ടാക്കുകയും ചെയ്യുന്നു.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ഫംഗ്ഷനുകൾ സൃഷ്ടിക്കുന്നു. 

തുടർന്ന് നമുക്ക് ഒരു ഫംഗ്ഷൻ നിർവചിക്കുകയും ആ ഫംഗ്ഷന്റെ പാരാമീറ്ററുകളും നിർവചിക്കുകയും ചെയ്യും. ഇവിടെ `search_courses` എന്ന ഒരു ഫംഗ്ഷൻ മാത്രമാണ് ഉപയോഗിക്കുന്നത്, പക്ഷേ നിങ്ങൾക്ക് എത്ര ഫംഗ്ഷനുകളും സൃഷ്ടിക്കാമെന്ന് മനസ്സിലാക്കാം.

**പ്രധാനമാണ്** : ഫംഗ്ഷനുകൾ LLM-ന് സിസ്റ്റം സന്ദേശത്തിനകത്തിൽ ഉൾപ്പെടുത്തപ്പെടുകയും നിങ്ങൾക്ക് ലഭ്യമായ ടോക്കണുകളുടെ എണ്ണം ഇതിൽ ഉൾപ്പെടുകയും ചെയ്യും. 


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**വ്യാഖ്യാനങ്ങൾ** 

`name` - വിളിക്കേണ്ട ഫംഗ്ഷന്റെ പേര്. 

`description` - ഫംഗ്ഷൻ എങ്ങനെ പ്രവർത്തിക്കുന്നു എന്നതിന്‍റെ വിവരണം. ഇവിടെ സ്പഷ്ടവും വ്യക്തമായും പറയുന്നത് പ്രധാനമാണ് 

`parameters` - മോഡൽ അതിന്‍റെ പ്രതികരണത്തിൽ സൃഷ്ടിക്കാൻ വേണ്ട മൂല്യങ്ങളുടെയും ഫോർമാറ്റിന്റെയും പട്ടിക 


`type` - സവിശേഷതകൾ സൂക്ഷിക്കേണ്ട ഡാറ്റാ തരം 

`properties` - മോഡൽ അതിന്‍റെ പ്രതികരണത്തിനായി ഉപയോഗിക്കുന്ന പ്രത്യേക മൂല്യങ്ങളുടെ പട്ടിക 


`name` - മോഡൽ അതിന്‍റെ ഫോർമാറ്റ് ചെയ്ത പ്രതികരണത്തിൽ ഉപയോഗിക്കുന്ന സവിശേഷതയുടെ പേര് 

`type` - ഈ സവിശേഷതയുടെ ഡാറ്റാ തരം 

`description` - പ്രത്യേക സവിശേഷതയുടെ വിവരണം 


**ഐച്ഛികം**

`required` - ഫംഗ്ഷൻ കോൾ പൂർത്തിയാക്കാൻ ആവശ്യമായ സവിശേഷത 


### ഫംഗ്ഷൻ കോൾ നിർമാണം  
ഒരു ഫംഗ്ഷൻ നിർവചിച്ചശേഷം, ഇപ്പോൾ അതിനെ ചാറ്റ് പൂരിപ്പിക്കൽ API യിൽ കോൾ ചെയ്യുന്നതിൽ ഉൾപ്പെടുത്തണം. ഞങ്ങൾ ഇത് അഭ്യർത്ഥനയിൽ `functions` ചേർത്ത് ചെയ്യുന്നു. ഈ ഘട്ടത്തിൽ `functions=functions` ആകുന്നു.  

`function_call` ന് `auto` എന്ന ഓപ്ഷനും ഉണ്ട്. ഇതിന് അർത്ഥം, ഉപയോഗക്കാരന്റെ സന്ദേശം അനുസരിച്ച് ഏത് ഫംഗ്ഷൻ കോൾ ചെയ്യണമെന്നും LLM തീരുമാനിക്കും, നമുക്ക് അതു തന്നെ നിശ്ചയിക്കേണ്ടതില്ല.  


```python
response = client.responses.create(model=deployment, 
                                        input=messages,
                                        tools=functions, 
                                        tool_choice="auto",
                                        store=False) 

print(response.output)
```

## 3.ഫംഗ്ഷൻ കോൾസുകളെ ഒരു ആപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കൽ. 


LLM ൽ നിന്ന് ഫോം ചെയ്ത പ്രതികരണം പരിശോധിച്ച ശേഷം, ഇപ്പോൾ നാം ഇതിനെ ഒരു ആപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കാം. 

### പ്രവാഹം നിയന്ത്രിക്കൽ 

ഇതിനെ നമ്മുടെ ആപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കാൻ, അടുത്ത നടപടികൾ സ്വീകരിക്കാം: 

ആദ്യമേ, ഓപ്പൺ എഐ സർവീസുകൾക്ക് കോൾ അയച്ച് സന്ദേശം `response_message` എന്ന വേരിയബിൾ ആയി സംഭരിക്കാം. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ഇപ്പോള്‍ നാം കോഴ്‌സുകളുടെ പട്ടിക ലഭിക്കാൻ Microsoft Learn API വിളിക്കുന്ന ഫംഗ്ഷൻ നിർവ്വചിക്കും: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

മികച്ച പ്രയോഗമായിട്ട്, മോഡൽ ഒരു ഫംഗ്ഷൻ വിളിക്കാൻ ആഗ്രഹിക്കുന്നുവോ എന്ന് നമുക്ക് ശേഷം നോക്കും. അതിന് ശേഷം, ലഭ്യമായ ഫംഗ്ഷനുകളിൽ ഒന്നിനെ സൃഷ്ടിച്ച് വിളിക്കപ്പെട്ട ഫംഗ്ഷനുമായി പൊരുത്തപ്പെടുത്തും. 
പിന്നീട്, ഫംഗ്ഷനിലെ Arguments എടുത്ത് അവ LLMയുടെ Arguments-ുമായി മാപ്പ് ചെയ്യും.

അവസാനം, ഫംഗ്ഷൻ കോൾ മെസേജും `search_courses` മെസേജ് തിരിച്ചു നൽകിയ മൂല്യങ്ങളും ചേർക്കും. ഇത് LLM-ന് ആവശ്യമായ എല്ലാ വിവരങ്ങളും ഉപയോക്താവിന് സ്വാഭാവിക ഭാഷ ഉപയോഗിച്ച് പ്രതികരിക്കാൻ നൽകുന്നു. 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ഇപ്പോൾ ഞങ്ങൾ അപ്‌ഡേറ്റ് ചെയ്ത സന്ദേശം LLM-യ്ക്ക് അയയ്ക്കും, അതിനാൽ നമുക്ക് API JSON ഫോർമാറ്റിലുള്ള പ്രതികരണത്തിന് പകരം സ്വാഭാവിക ഭാഷയിലെ പ്രതികരണം ലഭിക്കും.
