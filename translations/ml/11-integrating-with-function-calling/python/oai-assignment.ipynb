- OpenAI സേവനം ഉപയോഗിച്ച് ഫംഗ്ഷന്‍ کال്ല് ക്രമീകരിക്കുക 
- നിങ്ങളുടെ അപ്ലിക്കേഷന്റെ ഉപയോഗ കാര്യത്തിന് അനുയോജ്യമായ ഫംഗ്ഷന്‍ کال്ലുകൾ രൂപകല്‍പ്പന ചെയ്യുക 


## ഫംഗ്ഷൻ കോൾസ് മനസ്സിലാക്കുന്നു 

ഈ പാഠത്തിന് വേണ്ടി, നമ്മുടെ എഡ്യൂക്കേഷൻ സ്റ്റാർട്ടപ്പിന് ഒരു ഫീച്ചർ നിർമ്മിക്കാനാണ് നമ്മൾ ആഗ്രഹിക്കുന്നത്, ഇത് ഉപയോക്താക്കൾക്ക് ടെക്നിക്കൽ കോഴ്സുകൾ കണ്ടെത്താൻ ഒരു ചാറ്റ്ബോട്ട് ഉപയോഗിക്കാനാകും. അവരുടെ നിബദ്ധതാ നില, നിലവിലെ റോൾ, மற்றும் താൽപ്പര്യമുള്ള ടെക്നോളജി അനുസരിച്ച് കോഴ്സുകൾ ഞങ്ങൾ ശിപാർസു ചെയ്യുമെന്നും. 

ഇത് പൂർത്തിയാക്കാൻ നമ്മൾ താഴെ പറയുന്ന സംയോജനം ഉപയോഗിക്കും: 
 - ഉപയോക്താവിനായി ഒരു ചാറ്റ് അനുഭവം സൃഷ്ടിക്കാൻ `OpenAI`
 - ഉപയോക്താവിന്റെ അഭ്യർത്ഥന അടിസ്ഥാനമാക്കി കോഴ്സുകളെ കണ്ടെത്താൻ സഹായിക്കാൻ `Microsoft Learn Catalog API`
 - ഉപയോക്താവിന്റെ ചോദ്യം സ്വീകരിച്ച് API അഭ്യർത്ഥന നടത്താനുള്ള ഫംഗ്ഷനിലേക്ക് അയക്കാൻ `Function Calling`

ആരംഭിക്കാൻ, ആദ്യം നാം ഫംഗ്ഷൻ കോൾ ചെയ്യാൻ എന്തുകൊണ്ടാണ് ആഗ്രഹിക്കുന്നത് എന്ന് നോക്കാം: 


### ഫംഗ്ഷൻ കോളിംഗ് എന്തിന് 

ഈ കോഴ്സിലെ മറ്റ് ഏതു പാഠവും നിങ്ങൾ പൂർത്തിയാക്കി കഴിഞ്ഞാൽ, വലിയ ഭാഷാ മാതൃകകൾ (LLMs) ഉപയോഗിക്കുന്നതിന്റെ ശക്തി നിങ്ങൾക്ക് മനസ്സിലാകുമെന്ന് തോന്നുന്നു. അതോടൊപ്പം അവയുടെ ചില പരിമിതികളും നിങ്ങൾക്ക് കാണാൻ സാധിക്കുമെന്നും പ്രതീക്ഷിക്കുന്നു.

ഫംഗ്ഷൻ കോളിംഗ് ഒരു OpenAI സർവീസ് സവിശേഷതയാണ്, താഴെ പറയുന്ന വെല്ലുവിളികൾ പരിഹരിക്കാൻ രൂപകൽപ്പന ചെയ്തത്:

അശ്രദ്ധയുള്ള പ്രതികരണ രൂപകൽപ്പന:
- ഫംഗ്ഷൻ കോളിംഗിന് മുൻപ്, വലിയ ഭാഷാ മാതൃകയിൽ നിന്നുള്ള പ്രതികരണങ്ങൾ അസംരചനയും അശ്രദ്ധവും ആയിരുന്നു. ഡെവലപ്പർമാർ പ്രതികരണത്തിലെ ഓരോ വ്യത്യാസവും കൈകാര്യംചെയ്യാൻ സങ്കീർണ്ണമായ പരിശോധനാ കോഡ് എഴുതേണ്ടി വന്നിരുന്നു.

പുറമെ ഉള്ള ഡാറ്റയുമായി സാദുലഭമായ സംയോജനം:
- ഈ സവിശേഷതയ്ക്ക് മുൻപ്, ഒരു ആപ്ലിക്കേഷന്റെ മറ്റ് ഭാഗങ്ങളിൽനിന്നുള്ള ഡാറ്റ ഒരു ചാറ്റ് ഉൾക്കാഴ്ചയിലേക്കു ഉൾപ്പെടുത്തുന്നതിനുള്ള പ്രയാസം ഉണ്ടായിരുന്നു.

പ്രതികരണ ഫോർമാറ്റുകൾ സാധാരണരൂപത്തിലാക്കുകയും പുറമേ ഉള്ള ഡാറ്റയുമായി നിഗമന സൗകര്യം നൽകുകയും ചെയ്യുക മൂലം, ഫംഗ്ഷൻ കോളിംഗ് വികസനം ലളിതമാക്കുകയും അധിക പരിശോധനാ ലജിക് വേണ്ടിവരാതിരിക്കുകയുമാണ്.

ഉപഭോക്താക്കൾക്ക് "സ്റ്റോക്ക്ഹോംൽ മുഖേന നിലവിലെ കാലാവസ്ഥ എന്താണ്?" പോലുള്ള ചോദ്യങ്ങൾക്ക് ഉത്തരം ലഭിക്കാതെതിരുന്നു. കാരണം മാതൃകകൾ പരിശീലിപ്പിക്കപ്പെട്ട സമയത്തുവരെ ഉള്ള ഡാറ്റയ്ക്ക് മാത്രം പരിമിതമായിരുന്നു.

ഈ പ്രശ്നം വ്യക്തമായ ഉദാഹരണം നോക്കാം:

ശരിയായ പാഠം നിർദ്ദേശിക്കാൻ വിദ്യാർത്ഥികളുടെ ഡാറ്റാബേസ് രൂപകല്പന ചെയ്യണമെങ്കിൽ എന്തായിരിക്കും എന്നത് നമുക്ക് കാണാം. താഴെ രണ്ടു വിദ്യാർത്ഥികളുടെ വിവരങ്ങൾ വളരെ സമാനമാണ് ഉള്ളടക്ക açısından.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ഡാറ്റ പാഴ്‌സുചെയ്യാൻ ഇതു് ഒരു LLM ക്ക് അയയ്ക്കാൻ ഞങ്ങൾ ആഗ്രഹിക്കുന്നു. പിന്നീട് ഇത് നമ്മുടെ ആപ്ലിക്കേഷനിൽ API-യിലോ ഡാറ്റാബേസിൽ സൂക്ഷിക്കാനോ ഉപയോഗിക്കാം.

നാം LLM-നു് ഞങ്ങൾക്ക് ആഗ്രഹമുള്ള വിവരങ്ങൾ ഏതാണ് എന്ന് നിര്‍ദേശം നൽകാൻ രണ്ട് സമാനമായ പ്രോംപ്റ്റുകൾ ഉണ്ടാക്കാം:


ഞങ്ങൾ ഇത് ഒരു LLM-க்கு അയച്ചു നമ്മുടെ ഉൽപ്പന്നത്തിന് പ്രധാനപ്പെട്ട ഭാഗങ്ങളെ പാഴ്സുചെയ്യാൻ ആഗ്രഹിക്കുന്നു. അതിനാൽ LLM-നെ നിർദ്ദേശിക്കാൻ രണ്ട് സമാനമായ പ്രോംപ്റ്റുകൾ സൃഷ്ടിക്കാം: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ഈ രണ്ട് പ്രോംപ്റ്റുകളും സൃഷ്ടിച്ച ശേഷം, അവയെ LLM-க்கு `client.responses.create` ഉപയോഗിച്ച് അയക്കും.  പ്രോംപ്റ്റ് `input` എന്ന വേരിയബിളിൽ സൂക്ഷിക്കുകയും റോളിനെ `user` ആയി നിർദ്ദിശിക്കുകയും ചെയ്യും. ഇത് ഒരു ഉപയോക്താവിന്റെ സന്ദേശം ചാറ്റ്‌ബോട്ടിലേക്ക് എഴുതപ്പെടുന്നത് അതേപോലെ അനുകരിക്കാനായാണ്. 



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-5-mini"
```

ഇപ്പോൾ നാം ഇരുവിധ അഭ്യർത്ഥനകളും LLM-കു મોકച്ച് ലഭിക്കുന്ന പ്രതികരണത്തെ പരിശോധിക്കാം.


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


പ്രാമ്പ്റ്റുകളും വിവരണങ്ങളും സമാനമായിരുന്നാലും, `Grades` പ്രോപ്പർട്ടി വിവിധ ഫോർമാറ്റുകളിലായി കിട്ടാൻ കഴിയുന്നുണ്ട്.

മുകളിലുള്ള സെൽ മടങ്ങിമടങ്ങി ഓടിപ്പായാൽ, ഫോർമാറ്റ് `3.7` അല്ലെങ്കിൽ `3.7 GPA` ആകാൻ സാധ്യതയുണ്ട്.

ഇതു സംഭവിക്കുന്നത് LLM എഴുതി നൽകിയ പ്രാമ്പ്റ്റിന്റെ ആകാരരഹിത ഡാറ്റ സ്വീകരിച്ച് അതും ആകാരരഹിത ഡാറ്റയായി തിരിച്ച് നൽകുന്നതിനാൽ ആണ്. നമുക്ക് ഈ ഡാറ്റ സംഭരിച്ചു വയ്ക്കുമ്പോഴോ ഉപയോഗിക്കുമ്പോഴോ നെഞ്ചുതുമ്പി അറിയാൻ ഒരു ഘടിത ഫോർമാറ്റ് വേണം.

ഫംഗ്ഷണൽ കോളിംഗ് ഉപയോഗിച്ച് നമുക്ക് ഘടിത ഡാറ്റ തിരിച്ചുകൊണ്ടുവരാൻ സാധിക്കും. ഫംഗ്ഷൻ കോളിംഗ് ഉപയോഗിക്കുമ്പോൾ, LLM യഥാർത്ഥത്തിൽ ഏതൊരു ഫംഗ്ഷനും വിളിക്കുകയോ ഓടിക്കുകയോ ചെയ്യാറില്ല. അതിന് പകരം, LLM അതിന്റെ മറുപടികൾക്ക് പാലിക്കാനുള്ള ഘടന നാം സൃഷ്ടിക്കുന്നു. പിന്നെ ആ ഘടിത മറുപടികൾ ഉപയോഗിച്ചുകൊണ്ട്, നമ്മൾ നമ്മുടെ അപ്ലിക്കേഷനുകളിൽ ഏത് ഫംഗ്ഷൻ ഓടിക്കണമെന്ന് അറിയുന്നു.

ചാറ്റ്ബോട്ടുകൾ ഉപയോക്താക്കളുടെ ചോദ്യങ്ങൾക്ക് ഉത്തരം നൽകുന്നതിൽ മികച്ചവയാണ്. ഫങ്ഷൻ കോളിംഗ് ഉപയോഗിച്ച്, ചാറ്റ്ബോട്ടുകൾ ഉപയോക്താക്കളുടെ സന്ദേശങ്ങൾ ഉപയോഗിച്ച് ചില tácകൾ പൂര്‍ത്തിയാക്കാം. ഉദാഹരണത്തിന്, ഒരു വിദ്യാർത്ഥി ചാറ്റ്ബോട്ടിന് "ഈ വിഷയം കുറിച്ച് കൂടുതൽ സഹായം ആവശ്യമുണ്ടെന്ന് എന്റെ അധ്യാപക에게 ഇമെയിൽ അയയ്ക്കൂ" എന്ന് പറയാം. ഇതിന് `send_email(to: string, body: string)` എന്ന ഫങ്ഷൻ കോളുചെയ്യാം  


**API അല്ലെങ്കിൽ ഡേറ്റാബേസ് ക്വറികൾ സൃഷ്ടിക്കൽ**

**സംരചിത ഡാറ്റ സൃഷ്ടിക്കൽ**

## 2. നിങ്ങളുടെ ആദ്യ ഫങ്ക്ഷൻ കോൾ സൃഷ്ടിക്കൽ 

ഒരു ഫങ്ക്ഷൻ കോൾ സൃഷ്ടിക്കുന്ന പ്രക്രിയയിൽ 3 പ്രധാന പടികൾ ഉൾപ്പെടുന്നു: 
1. നിങ്ങളുടെ ഫങ്ക്ഷനുകളുടെ ലിസ്റ്റും ഒരു ഉപയോക്തൃ സന്ദേശവും ഉൾപ്പെടുത്തി ചാറ്റ് കമ്പ്ളീഷൻസ് API കാൾ ചെയ്യുക 
2. ഫംഗ്ഷൻ അല്ലെങ്കിൽ API കോളിന് പ്രവർത്തനം നടത്താൻ മോഡലിന്റെ പ്രതികരണം വായിക്കുക 
3. ഉപയോക്താവിനായുള്ള ഒരു പ്രതികരണം സൃഷ്ടിക്കാൻ നിങ്ങളുടെ ഫങ്ക്ഷനിൽ നിന്നുള്ള പ്രതികരണത്തിനെ ഉപയോഗിച്ച് ചാറ്റ് കമ്പ്ളീഷൻസ് API-യെ മറ്റൊരു കാൾ ചെയ്യുക.

`role` എന്നത് `system` (നിയമങ്ങൾ സൃഷ്ടിക്കുന്നു), `assistant` ( മോഡൽ) അല്ലെങ്കിൽ `user` (അവസാന ഉപയോക്താവ്) ആയിരിക്കാൻ കഴിയും. function calling നു വേണ്ടി, നാം ഇത് `user` ആയി നിയോഗിക്കും, കൂടാതെ ഒരു ഉദാഹരണ ചോദ്യം നൽകും.


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ഫംഗ്ഷനുകൾ സൃഷ്ടിക്കൽ. 

తరువాత, നാം ഒരു ഫംഗ്ഷൻ നിർവചിക്കുകയും ആ ഫംഗ്ഷന്റെ പാരാമീറ്ററുകളും നിർവചിക്കുകയും ചെയ്യും. ഞങ്ങൾ ഇവിടെ `search_courses` എന്ന ഒറ്റ ഫംഗ്ഷനുകൾ മാത്രം ഉപയോഗിക്കും, പക്ഷേ നിങ്ങൾ പല ഫംഗ്ഷനുകളും സൃഷ്ടിക്കാവുന്നതാണ്.

**പ്രധാനമായത്**: ഫംഗ്ഷനുകൾ LLM-നെയുള്ള സിസ്റ്റം സന്ദേശത്തിൽ ഉൾപ്പെടുത്തിയിരിക്കും, കൂടാതെ ഇത് നിങ്ങൾക്കുള്ള ലഭ്യമായ ടോക്കനുകളുടെ അളവിൽ ഉൾപ്പെടുത്തിയിരിക്കും.


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**นิยามം** 

ഫംഗ്ഷന് നിർവചന ഘടനയിൽ പല തലങ്ങളുണ്ട്, ഓരോതിലും തങ്ങളുടെ സ്വന്തം സ്വഭാവങ്ങൾ ഉണ്ട്. സങ്കീർണ്ണ ഘടനയുടെ ഒരു വിഭജനം ഇവിടെ കൊടുക്കുന്നു:

**ഉത്തരതല ഫംഗ്ഷൻ സ്വഭാവങ്ങൾ:**

`name` - നമ്മള് വിളിക്കാൻ ആഗ്രഹിക്കുന്ന ഫംഗ്ഷനിന്റെ പേര്.

`description` - ഫംഗ്ഷൻ എങ്ങനെ പ്രവർത്തിക്കുന്നു എന്നതിനെക്കുറിച്ചുള്ള വിവരണം. ഇതിൽ വ്യക്തവും വ്യക്തമുമായിരിക്കണം.

`parameters` - മോഡൽ പ്രധാനം ചെയ്യേണ്ട മൂല്യങ്ങളും ഫോർമാറ്റും ഉള്ള പട്ടിക.

**പാരാമീറ്റർ ഒബ്ജക്റ്റിന്റെ സ്വഭാവങ്ങൾ:**

`type` - പാരാമീറ്ററുകളുടെ ഒബ്ജക്റ്റിന്റെ ഡാറ്റ(TYPE) തരം (പൊതു "object")

`properties` - മോഡൽ പ്രതികരിക്കാൻ ഉപയോഗിക്കുന്ന നിശ്ചിത മൂല്യങ്ങളുടെ പട്ടിക.

**തനതനായ പാരാമീറ്റർ സ്വഭാവങ്ങൾ:**

`name` - പ്രോപ്പർട്ടി കീ വഴി വ്യക്തമാക്കപ്പെടുന്നു (ഉദാഹരണം, "role", "product", "level")

`type` - ഈ പ്രത്യേക പാരാമീറ്ററിന്റെ ഡാറ്റ തരവും (ഉദാഹരണം, "string", "number", "boolean")

`description` - പ്രത്യേക പാരാമീറ്ററിന്റെ വിവരണം.

**ഐച്ഛിക സ്വഭാവങ്ങൾ:**

`required` - ഫംഗ്ഷൻ വിളവ് പൂർത്തിയാക്കാൻ ഏത് പാരാമീറ്ററുകൾ ആവശ്യമാണെന്ന് പട്ടികവത്കരിക്കുന്ന ഒരു ശ്രേണി.


### ഫംഗ്ഷൻ കോളിംഗ് നടത്തൽ
ഒരു ഫംഗ്ഷൻ നിർവ്വചിച്ചശേഷം, അതിനെ Chat Completion API-യുടെ کالിൽ ഉൾപ്പെടുത്തേണ്ടതാണ്. ഞങ്ങൾ ഇത് `functions` ക്ക് റിക്വസ്റ്റ് കൊടുക്കുമ്പോൾ ചേർത്തുകൊണ്ട് ചെയ്യുന്നു. ഈ ഘട്ടത്തിൽ `functions=functions` ആണ് ഉപയോഗിക്കുന്നത്.

`function_call` നെ `auto` ആയി സജ്ജീകരിക്കാൻ ഒരു ഓപ്ഷനും ഉണ്ട്. അർത്ഥം, ഞങ്ങൾ തന്നെ ഏത് ഫംഗ്ഷൻ വിളിക്കണമെന്നു തിരഞ്ഞെടുത്ത് നൽകാതെ, ഉപയോക്തൃ സന്ദേശത്തിന്റെ അടിസ്ഥാനത്തിൽ LLM നു החלטి ചെയ്യാൻ അനുവദിക്കുക എന്നതാണ്.

## 3.ഫങ്ക്ഷൻ കോളുകൾ ഒരു അപ്ലിക്കേഷനിൽ സംയോജിപ്പിക്കുക. 


LLM-ലെ ഫോർമാറ്റു ചെയ്യപ്പെട്ട മറുപടി നാം പരിശോധിച്ച ശേഷം, ഇനിയെ ഇത് ഒരു അപ്ലിക്കേഷനിൽ സംയോജിപ്പിക്കാം. 

### ഫ്ലോ മാനേജ്മെന്റ് 

ഇത് നമ്മുടെ അപ്ലിക്കേഷനിൽ സംയോജിപ്പിക്കുന്നതിന്, താഴെക്കൊടുത്തിരിക്കുന്ന പടികളെടുക്കാം: 

ആദ്യം, OpenAI സേവനങ്ങളിലേക്ക് കോളു ചെയ്ത ശേഷം സന്ദേശം `response_message` എന്ന വ്യത്യാസ്രഹിതത്തിൽ സൂക്ഷിക്കാം. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ഇനി നാം കോഴ്‌സുകളുടെ പട്ടിക പ്രാപ്തമാക്കാൻ മൈക്രോസോഫ്റ്റ് ലേൺ API 호출 ചെയ്യാനുള്ള ഫംഗ്ഷൻ നിർവ്വചിക്കാം: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

ഉത്തമ പ്രയോഗമായി, മോഡൽ ഒരു ഫംഗ്ഷൻ വിളിക്കാൻ ആഗ്രഹിക്കുന്നുണ്ടോ എന്ന് നാം പരിശോധിക്കും. അതിനുശേഷം, ലഭ്യമായ ഫംഗ്ഷനുകളിൽ ഒരുത് സൃഷ്ടിച്ച് വിളിക്കുന്ന ഫംഗ്ഷനുമായി ഇത് പൊരുത്തപ്പെടുത്തും. 
പിന്നീട്, ഫംഗ്ഷനിലെ arguments എടുത്ത് LLM സർവ്വീസിൽ നിന്നുള്ള arguments ന് മാപ്പ് ചെയ്യും.

ഒടുവിൽ, ഫംഗ്ഷൻ കോൾ സന്ദേശവും `search_courses` സന്ദേശം തിരിച്ചയച്ച മൂല്യങ്ങളും ഞങ്ങൾ ചേർക്കും. ഇതു LLM ക്ക് വേണ്ടതെല്ലാം വിവരങ്ങൾ നൽകുന്നു
ഉപയോക്താവിനെ സ്വാഭാവിക ഭാഷയിൽ പ്രതികരിക്കാൻ. 


```python
# Check if the model wants to call a function
if tool_calls:
    tool_call = tool_calls[0]
    print("Recommended Function call:")
    print(tool_call.name)
    print()

    # Call the function. 
    function_name = tool_call.name

    available_functions = {
            "search_courses": search_courses,
    }
    function_to_call = available_functions[function_name] 

    function_args = json.loads(tool_call.arguments)
    function_response = function_to_call(**function_args)

    print("Output of function call:")
    print(function_response)
    print(type(function_response))


    # Add the model's function call item(s) and our function result to the conversation.
    # The Responses API represents tool results as `function_call_output` items.
    messages.extend(response.output)  # adding the model's function_call item(s)
    messages.append( # adding function response to messages
        {
            "type": "function_call_output",
            "call_id": tool_call.call_id,
            "output": function_response,
        }
    )
```

ഇപ്പോൾ നാം അപ്ഡേറ്റ് ചെയ്ത സന്ദേശം LLM-க்கு അയയ്ക്കുന്നു, ഇതിലൂടെ API JSON ഫോർമാറ്റിൽ ഉള്ള മറുപടി പകരം സ്വാഭാവിക ഭാഷയിലെ പ്രതികരണം സ്വീകരിക്കാനാകും.
