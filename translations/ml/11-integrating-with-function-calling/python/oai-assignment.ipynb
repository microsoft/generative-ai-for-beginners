- നിങ്ങളുടെ അപ്ലിക്കേഷന്റെ ഉപയോഗത്തിന് അനുയോജ്യമായ ഫംഗ്ഷൻ കോൾ ഡിസൈൻ ചെയ്യുക 


## ഫങ്ഷന്‍ കോളുകള്‍ മനസ്സിലാക്കുക

ഈ പാഠത്തിന് വേണ്ടി, നമ്മുടെ എജ്യൂക്കേഷന്‍ സ്റ്റാര്‍ട്ടപ്പിനു വേണ്ടി ഒരു ഫീച്ചർ നിർമ്മിക്കണമെന്ന് ആഗ്രഹിക്കുന്നു, അതിൽ ഉപയോക്താക്കൾക്ക് ടെക്‌നിക്കല്‍ കോഴ്സുകൾ കണ്ടെത്താൻ ചാറ്റ്ബോട്ട് ഉപയോഗിക്കാം. അവരുടെ താൽപര്യമുള്ള കഴിവ് നില, ഇപ്പോഴത്തെ റോളും സാങ്കേതികവിദ്യയും പരിഗണിച്ച് ഞങ്ങൾ കോഴ്സുകൾ ശിപാർശ ചെയ്യും.

ഇതു പൂർത്തിയാക്കുന്നതിന് നമ്മൾ സംയോജിപ്പിക്കും:
 - ഉപയോക്താവിന് ചാറ്റ് അനുഭവം സൃഷ്ടിക്കാൻ `OpenAI`
 - ഉപയോക്താവിന്റെ അഭ്യർത്ഥനയുടെ അടിസ്ഥാനത്തില്‍ കോഴ്സുകള്‍ കണ്ടെത്താന്‍ `Microsoft Learn Catalog API`
 - ഉപയോക്താവിന്റെ ചോദ്യം എടുക്കാനും API അഭ്യർത്ഥന നടത്താൻ ഫങ്ഷനിലേക്ക് അയയ്ക്കാനും `Function Calling`

തുടങ്ങാൻ, ആദ്യം നമുക്ക് function calling ഉപയോഗിക്കാൻ എങ്ങനെ ആവശ്യമാകും എന്ന് നോക്കാം:


### ഫംഗ്ഷൻ കോളിംഗ് എന്തുകൊണ്ട് 

ഈ കോഴ്സിലെ മറ്റ് പാഠങ്ങൾ നിങ്ങൾ പൂർത്തിയാക്കിയിട്ടുണ്ടെങ്കിൽ, നിങ്ങൾക്ക് വലിയ ഭാഷാ മോഡലുകൾ (LLMs) ഉപയോഗിക്കുന്നതിന്റെ ശക്തി മനസ്സിലാവും. അതുപോലെ അവരുടെ ചില പരിധികളും നിങ്ങൾക്ക് കാണാൻ കഴിയും എന്ന് പ്രതീക്ഷിക്കുന്നു. 

ഫംഗ്ഷൻ കോളിംഗ് ഓപ്പൺഎഐ സർവീസിന്റെ ഒരു ഫീച്ചറാണ്, താഴെ പറയുന്ന വെല്ലുവിളികളെ നേരിടാൻ രൂപകൽപ്പന ചെയ്തിരിക്കുന്നത്:

അസംഘടിത പ്രതികരണ ഫോർമാറ്റിംഗ്:
- ഫംഗ്ഷൻ കോളിംഗ് മുമ്പ്, വലിയ ഭാഷാ മോഡലിൽ നിന്ന് ലഭിക്കുന്ന പ്രതികരണങ്ങൾ അസംഘടിതവും അനിയമിതവുമായിരുന്നതായി. ഉൽപ്പന്നത്തിലെ ഓരോ വ്യത്യാസത്തിനും നിയന്ത്രണം നടത്താൻ ഡെവലപ്പർമാർക്ക് സങ്കീৰ্ণമായ സത്യാപന കോഡ് എഴുതേണ്ടിവന്നു.

ബാഹ്യ ഡേറ്റയുമായി പരിമിതമായ സംയോജനം:
- ഈ ഫീച്ചറിനു മുമ്പ്, ആപ്ലിക്കേഷന്റെ മറ്റ് ഭാഗങ്ങളിൽ നിന്നുള്ള ഡാറ്റ ചാറ്റ് സന്ദർഭത്തിൽ ചേര്‍ക്കുക കഠിനമായിരുന്നു.

പ്രതികരണ ഫോർമാറ്റുകൾ സ്റ്റാൻഡർഡൈസ് ചെയ്ത് ബാഹ്യ ഡേറ്റയുമായുള്ള സമഗ്രമായ സംയോജനം സാധ്യമാക്കുന്നതിലൂടെ, ഫംഗ്ഷൻ കോളിംഗ് ഡെവലപ്പ്മെന്റ് എളുപ്പമാക്കുകയും അധിക സത്യാപന लॉജിക് ആവശ്യകത കുറക്കുകയും ചെയ്യുന്നു.

ഉപയോക്താക്കൾക്ക് "സ്റ്റോക്ക്ഹോം-ൽ നിലവിലെ കാലാവസ്ഥ എന്താണ്?" പോലുള്ള ഉത്തരങ്ങൾ ലഭിക്കാറില്ല. കാരണം മോഡലുകൾ പരിശീലന സമയത്തുള്ള ഡേറ്റ്യാസമയഭേദങ്ങൾക്കു പോലും പരിമിതമായിരുന്നു. 

താഴെ കാണിക്കുന്ന ഉദാഹരണം ഈ പ്രശ്‌നം വിശദീകരിക്കുന്നു: 

വിദ്യാർത്ഥികളുടെ ഡാറ്റാബേസ് ഉണ്ടാക്കി, അവർക്ക് അനുയോജ്യമായ കോഴ്സ് നിർദ്ദേശിക്കാൻ എന്ന് സമ്പ്രദായമാക്കാൻ പോകുന്നു എന്ന് പറയാം. താഴെ രണ്ട് വിദ്യാർത്ഥികളുടെയും ഡാറ്റയിൽ കൂടുതലായി സമാനമായ രണ്ട് വിവരണങ്ങളുണ്ട്.


```python
student_1_description="Emily Johnson is a sophomore majoring in computer science at Duke University. She has a 3.7 GPA. Emily is an active member of the university's Chess Club and Debate Team. She hopes to pursue a career in software engineering after graduating."
 
student_2_description = "Michael Lee is a sophomore majoring in computer science at Stanford University. He has a 3.8 GPA. Michael is known for his programming skills and is an active member of the university's Robotics Club. He hopes to pursue a career in artificial intelligence after finishing his studies."
```


ഈ ഡാറ്റ പാഴ്സുചെയ്യാൻ LLM-ൽ ഇത് അയയ്ക്കാൻ ഞങ്ങൾ ആഗ്രഹിക്കുന്നു. പിന്നീട് ഇത് ഒരു API-യിലേക്കോ ഡാറ്റാബേസിൽ সংരക്ഷിക്കാനോ നമ്മുടെ അപ്ലിക്കേഷനിൽ ഉപയോഗിക്കാം.

ഞങ്ങൾ ആവേശപ്പെടുന്ന വിവരങ്ങൾ എന്തെല്ലാമാണെന്ന് LLM-നെ അറിയിക്കുന്ന രണ്ട് പകർപ്പ്_PROMPTS_ ഉണ്ടാക്കാം:


ഞങ്ങൾ ഇത് നമ്മുടെ ഉൽപ്പന്നത്തിനായി പ്രധാനപ്പെട്ട ഭാഗങ്ങൾ വിശകലനം ചെയ്യാൻ ഒരു LLM-ക്ക് അയയ്ക്കാൻ ആഗ്രഹിക്കുന്നു. അതിനാൽ LLM-നെ നിർദ്ദേശിക്കാൻ രണ്ട് തുല്യ പ്രോമ്പ്റ്റുകൾ സൃഷ്ടിക്കാം: 


```python
prompt1 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_1_description}
'''


prompt2 = f'''
Please extract the following information from the given text and return it as a JSON object:

name
major
school
grades
club

This is the body of text to extract the information from:
{student_2_description}
'''
```

ഈ രണ്ട് പ്രോംപ്റ്റുകളും സൃഷ്ടിച്ചശേഷം, `client.responses.create` ഉപയോഗിച്ച് അവ LLM-ന് അയയ്ക്കും.  പ്രോംപ്റ്റ് `input` വ്യത്യാസത്തിൽ സൂക്ഷിക്കുകയും റോളിനെ `user` ആയി നിശ്ചയിക്കുകയും ചെയ്യുന്നു. ഇത് ഒരു ഉപയോക്താവിനാൽ ചാറ്റ്ബോട്ടിലേക്ക് എഴുതുന്ന സന്ദേശം അനുകരിക്കാൻ ആണ്.



```python
import os
import json
from openai import OpenAI
from dotenv import load_dotenv
load_dotenv()

client = OpenAI()

deployment="gpt-4o-mini"
```

ഇപ്പോള്‍ നാം രണ്ടുപ്രശ്നങ്ങളും LLM-ന് അയച്ചു, ലഭിക്കുന്ന മറുപടി പരിശോധിക്കാം. 


```python
openai_response1 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt1}],
 store=False,
)
openai_response1.output_text 
```

```python
openai_response2 = client.responses.create(
 model=deployment,    
 input = [{'role': 'user', 'content': prompt2}],
 store=False,
)
openai_response2.output_text
```

```python
# Loading the response as a JSON object
json_response1 = json.loads(openai_response1.output_text)
json_response1
```

```python
# Loading the response as a JSON object
json_response2 = json.loads(openai_response2.output_text)
json_response2
```


പ്രോംപ്റ്റുകൾ ഒരുപോലെയും വിവരണങ്ങൾ സമാനവുമാണെങ്കിലും, ഞങ്ങൾ `Grades` പ്രോപ്പർട്ടി വ്യത്യസ്ത ഫോർമാറ്റുകളിൽ ലഭിക്കാം.

മുകളിൽ കൊടുത്ത സെൽ പലതവണ പ്രവർത്തിപ്പിച്ചാൽ, ഫോർമാറ്റ് `3.7` അല്ലെങ്കിൽ `3.7 GPA` ആയിരിക്കും.

എഴുത്തുപ്രോംപ്റ്റിന്റെ രൂപത്തിൽ LLM ഘടനയില്ലാത്ത ഡാറ്റ സ്വീകരിച്ച് ഘടനയില്ലാത്ത ഡാറ്റയും തിരിച്ച് നൽകുന്നതിനാലാണ് ഇത്. നാം ഈ ഡാറ്റ സംഭരിക്കുമ്പോഴും ഉപയോഗിക്കുമ്പോഴും എന്ത് പ്രതീക്ഷിക്കാമെന്നറിയാൻ ഘടനയുള്ള ഫോർമാറ്റ് വേണം.

ഫംക്ഷണൽ കോൾ ചെയ്യലിൽ ഉപയോഗിക്കുമ്പോൾ, ഘടനയുള്ള ഡാറ്റ തിരിച്ച് കിട്ടുമെന്ന് ഉറപ്പാക്കാം. ഫംക്ഷൻ കോൾ ചെയ്യുമ്പോൾ, LLM യഥാർത്ഥത്തിൽ ഫംക്ഷനുകൾ കോൾ ചെയ്യുകയോ ഓടിക്കുകയോ ചെയ്യാറില്ല. അതിന്റെ பதിലുകൾക്ക് LLM പാലിക്കാനുള്ള ഘടന നാം സൃഷ്‌ടിക്കുന്നു. തുടർന്ന് ആ ഘടനാപരമായ പ്രതികരണങ്ങൾ ഉപയോഗിച്ച് ഞങ്ങളുടെ അപ്ലിക്കേഷനുകളിൽ ഏത് ഫംക്ഷൻ നടത്തണമെന്ന് നാം അറിയുന്നു.

## 2. നിങ്ങളുടെ ആദ്യ ഫംഗ്ഷൻ കാൾ സൃഷ്‌ടിക്കൽ 

ഒരു ഫംഗ്ഷൻ കാൾ സൃഷ്‌ടിക്കുന്ന പ്രക്രിയയിൽ 3 പ്രധാന ഘട്ടങ്ങൾ ഉൾപ്പെടുന്നു: 
1. നിങ്ങളുടെ ഫംഗ്ഷനുകളുടെ പട്ടികയും ഒരു ഉപയോക്തൃ സന്ദേശവും ഉപയോഗിച്ച് ചാറ്റ് കോംപ്ലഷൻസ് API കൾ ചെയ്യുക 
2. പ്രവർത്തനം നടത്താനും ഫംഗ്ഷൻ അല്ലെങ്കിൽ API കാൾ നടപ്പിലാക്കാനും മോഡലിന്റെ പ്രതികരണം വായിക്കുക 
3. ഉപയോക്താവിന് പ്രതികരണം സൃഷ്‌ടിക്കാൻ ഉപയോഗിക്കുന്നതിനായി നിങ്ങളുടെ ഫംഗ്ഷനിൽ നിന്നുള്ള പ്രതികരണത്തോടെ ചാറ്റ് കോംപ്ലഷൻസ് API വീണ്ടും വിളിക്കുക.

### ഒരു ഫംഗ്ഷൻ കോൾ എന്നതിന്റെ ഘടകങ്ങൾ 

#### ഉപഭോക്തൃതിൻറെ ഇൻപുട്ട് 

ആദ്യത്തെ ഘട്ടം ഉപയോക്തൃ സന്ദേശം സൃഷ്ടിക്കുകയാണ്. ഇത് ടെക്സ്റ്റ് ഇൻപുട്ടിന്റെ മൂല്യം എടുത്ത് ഡൈനാമിക്കായി നിയോഗിക്കാവുന്നതും, അല്ലെങ്കിൽ ഇത്തിക്കുന്ന ഒരു മൂല്യം നിങ്ങൾ ഇവിടെ നിയോഗിക്കാവുന്നതുമായിരിക്കും. നിങ്ങൾക്ക് Chat Completions API ഉപയോഗിച്ച് ആദ്യമായിരിക്കും പ്രവർത്തിക്കുന്നത് എങ്കിൽ, సందേശത്തിന്റെ `role`യും `content`ഉം നിർവചിക്കേണ്ടതുണ്ട്. 

`role` എന്ന് പറയുന്നത് `system` (നിയമങ്ങൾ സൃഷ്ടിക്കൽ), `assistant` (മോഡൽ) അല്ലെങ്കിൽ `user` (അവസാന ഉപയോക്താവ്) ആയിരിക്കും. ഫംഗ്ഷൻ കോൾ ചെയ്യുന്നതിന്, നാം ഇത് `user` ആയി ക്രമീകരിക്കുകയും ഒരു ഉദാഹരണ ചോദ്യമുണ്ടാക്കുകയും ചെയ്യും. 


```python
messages= [ {"role": "user", "content": "Find me a good course for a beginner student to learn Azure."} ]
```

### ഫംഗ്ഷനുകൾ സൃഷ്ടിക്കുന്നത്. 

അടുത്തതായി നാം ഒരു ഫംഗ്ഷനും ആ ഫംഗ്ഷന്റെ പാരാമീറ്ററുകളും നിർവചിക്കും. ഇവിടെ `search_courses` എന്ന ഒരു ഫംഗ്ഷനെ മാത്രം ഉപയോഗിക്കാം, പക്ഷേ നിങ്ങൾക്ക് ഒന്നിലധികം ഫംഗ്ഷനുകൾ സൃഷ്ടിക്കാം.

**പ്രധാനമാം**: ഫംഗ്ഷനുകൾ സിസ്റ്റം സന്ദേശത്തിൽ LLM-ന്റെ ഭാഗമായും ഉൾപ്പെടുന്നു, ഇത് നിങ്ങൾക്ക് ലഭിക്കുന്ന ടോക്കണുകളുടെ കണക്കിൽ ഉൾപ്പെടും.


```python
# The Responses API uses a flat tool format: name/description/parameters at the top level
functions = [
   {
      "type":"function",
      "name":"search_courses",
      "description":"Retrieves courses from the search index based on the parameters provided",
      "parameters":{
         "type":"object",
         "properties":{
            "role":{
               "type":"string",
               "description":"The role of the learner (i.e. developer, data scientist, student, etc.)"
            },
            "product":{
               "type":"string",
               "description":"The product that the lesson is covering (i.e. Azure, Power BI, etc.)"
            },
            "level":{
               "type":"string",
               "description":"The level of experience the learner has prior to taking the course (i.e. beginner, intermediate, advanced)"
            }
         },
         "required":[
            "role"
         ]
      }
   }
]
```

**നിർവചനങ്ങൾ**

`description` - ഫംഗ്ഷൻ എങ്ങനെ പ്രവർത്തിക്കുന്നു എന്നതിന്റെ വിവരണം. ഇവിടെ സ്പഷ്ടവും സുതാര്യമുള്ളതുമായിരിക്കുന്നത് പ്രധാനമാണ് 

`parameters` - മോഡൽ അതിന്റെ പ്രതികരണത്തിൽ ഉൽപാദിപ്പിക്കണമെന്ന് നിങ്ങൾ ആഗ്രഹിക്കുന്ന മൂല്യങ്ങളുടെയും ഫോർമാറ്റിന്റെയും പട്ടിക 

**പരാമീറ്റർ ഓബ്ജക്റ്റ് പ്രോപ്പർട്ടികൾ:**

`type` - പരാമീറ്റർ ഓബ്ജക്റ്റിന്റെ ഡേറ്റാ ടൈപ്പ് (സാധാരണയായി "object") 

`properties` - മോഡൽ അതിന്റെ പ്രതികരണത്തിന് ഉപയോഗിക്കും എന്ന് നിശ്ചയിച്ച പ്രത്യേക മൂല്യങ്ങളുടെ പട്ടിക

`description` - പ്രത്യേക പരാമീറ്ററിന്റെ വിവരണം 

**ഐച്ഛിക പ്രോപ്പർട്ടികൾ:**

`required` - ഫംഗ്ഷൻ വിളി പൂർത്തിയാക്കാൻ ആവശ്യമായ പരാമീറ്ററുകളുടെ പട്ടിക

## 3.ഫംഗ്ഷൻ കോൾസുകൾ ഒരു അപ്ലിക്കേഷനിൽ സംയോജിപ്പിക്കൽ. 


LLM നിന്നുള്ള ഫോർമാറ്റ് ചെയ്ത മറുപടി ഞങ്ങൾ പരീക്ഷിച്ച ശേഷം, ഇപ്പോൾ ഇത് ഒരു അപ്ലിക്കേഷനിൽ സംയോജിപ്പിക്കാം. 

### ഫ്ലോ മാനേജ്മെന്റ് 

ഇത് നമ്മുടെ അപ്ലിക്കേഷനിലേക്ക് സംയോജിപ്പിക്കാൻ, നമുക്ക് താഴെ പറയുന്നപോലെ ചുവടുകളെ സ്വീകരിക്കാം: 

ആദ്യമായി, OpenAI സേവനങ്ങളെ വിളിച്ച്, സന്ദേശം `response_message` എന്ന വേരിയബിളിൽ സൂക്ഷിക്കാം. 


```python
# Extract the function call items from the response output
tool_calls = [item for item in response.output if item.type == "function_call"]
```

ഇപ്പോൾ നാം കോഴ്സുകളുടെ പട്ടിക ലഭിക്കാൻ Microsoft Learn API ആഹ്വാനം ചെയ്യുന്ന ഫംഗ്ഷൻ നിർവചിക്കാം: 


```python
import requests

def search_courses(role, product, level):
    url = "https://learn.microsoft.com/api/catalog/"
    params = {
        "role": role,
        "product": product,
        "level": level
    }
    response = requests.get(url, params=params)
    modules = response.json()["modules"]
    results = []
    for module in modules[:5]:
        title = module["title"]
        url = module["url"]
        results.append({"title": title, "url": url})
    return str(results)

```

പിന്നീട്, ഫംഗ്ഷന്റെ_arguments_ എടുത്തെത്തിച്ചു അവ LLM-യിലെ_arguments_-നൊപ്പം മാപ്പ് ചെയ്യും.

അവസാനമായി, ഫംഗ്ഷൻ കോൾ സന്ദേശവും `search_courses` സന്ദേശം വഴി തിരികെ ലഭിച്ച മൂല്യങ്ങളും ചേർക്കും. ഇതിലൂടെ LLM-ന് ഉപയോക്താവിന് പ്രകൃതിഭാഷ ഉപയോഗിച്ച് പ്രതികരിക്കാൻ ആവശ്യമായ എല്ലാ വിവരങ്ങളും ലഭിക്കും.
