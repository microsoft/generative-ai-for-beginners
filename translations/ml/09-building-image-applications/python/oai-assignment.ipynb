## DALL-Eയും Midjourneyയും എങ്ങനെ പ്രവർത്തിക്കുന്നു 

ആദ്യം [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst) നോക്കാം. DALL-E ഒരു ട്രാൻസ്ഫോർമർ ആർക്കിടെക്ചർ അടിസ്ഥാനമാക്കിയുള്ള *ആറടുത്ത് സ്വയം നിര്‍ദേശിക്കുന്ന ട്രാൻസ്ഫോർമർ* ആയ ഒരു ജനറേറ്റീവ് AI മോഡലാണ്.  

*ആറടുത്ത് സ്വയം നിര്‍ദേശിക്കുന്ന ട്രാൻസ്ഫോർമർ* ഒരു മോഡലിൽ നിന്ന് ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്നുള്ള ചിത്രങ്ങൾ എങ്ങനെ സൃഷ്ടിക്കാമെന്ന് നിർവചിക്കുന്നു, ഇത് ഒരു പിക്സലും ഒരു സമയം സൃഷ്ടിച്ച് പിന്നീട് സൃഷ്ടിച്ച പിക്സലുകൾ ഉപയോഗിച്ച് അടുത്ത പിക്സൽ സൃഷ്ടിക്കുന്നു. നിരവധി നൂറൽ നെറ്റ്‌വർക് ലെയറുകളിലൂടെ കടന്നുപോയി ചിത്രം പൂർണ്ണമാകുന്നതുവരെ.  

ഈ പ്രക്രിയയിലൂടെ, DALL-E ചിത്രത്തിൽ ഉള്ള അറ്റ്രിബ്യൂട്ടുകൾ, വസ്തുക്കൾ, സവിശേഷതകൾ തുടങ്ങിയവ നിയന്ത്രിക്കുന്നു. എന്നാൽ DALL-E 2, 3 ന്റെ സൃഷ്ടിച്ച ചിത്രത്തിന് കൂടുതൽ നിയന്ത്രണം ഉണ്ട്. 


## നിങ്ങളുടെ ആദ്യ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കൽ

അപ്പോൾ ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കാൻ എന്താണ് ആവശ്യം? നിങ്ങൾക്ക് താഴെയുള്ള ലൈബ്രറികൾ വേണം:

- **python-dotenv**, ഈ ലൈബ്രറി ഉപയോഗിച്ച് നിങ്ങളുടെ രഹസ്യങ്ങൾ കോഡിന് മെയ്‌ക്കാതെ *.env* ഫയലിൽ സൂക്ഷിക്കാൻ ശക്തമായി ശുപാർശ ചെയ്യപ്പെടുന്നു.
- **openai**, OpenAI API യുമായി സംവദിക്കാൻ നിങ്ങൾ ഉപയോഗിക്കുന്നത് ഈ ലൈബ്രറിയാണ്.
- **pillow**, Python-ൽ تصويرങ്ങൾ ഉപയോഗിക്കാൻ.
- **requests**, HTTP അഭ്യർഥനകൾ ചെയ്യാൻ സഹായിക്കുന്നു.


1. താഴെയുള്ള ഉള്ളടക്കമുള്ള *.env* ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. മുകളിൽ പറഞ്ഞ ലൈബ്രറികളെ *requirements.txt* എന്ന ഫയലിൽ ഇങ്ങനെ സമാഹരിക്കുക:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. തുടർന്ന്, വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിക്കുകയും ലൈബ്രറികൾ ഇൻസ്റ്റാൾ ചെയ്യുകയും ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows-ന്, നിങ്ങളുടെ വിർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിച്ച് സജീവമാക്കാൻ താഴെ കാണുന്ന കമാൻഡുകൾ ഉപയോഗിക്കുക:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന ഫയലിൽ താഴെ കാണുന്ന കോഡ് ചേർക്കുക:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()

    # OpenAI объекты സൃഷ്ടിക്കുക
    client = openai.OpenAI()


    try:
        # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ചിത്രം സൃഷ്ടിക്കുക
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് വാചകം ഇവിടെ ചേർക്കുക
            size='1024x1024',
            n=1
        )
        # സ്റ്റോർ ചെയ്തിരിക്കുന്ന ചിത്രത്തിനായുള്ള ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')

        # ഡയറക്ടറി നിലവിലില്ലെങ്കിൽ സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ഇമേജ് പാത്ത് ഇൻഷ്യലൈസ് ചെയ്യുക (ഫയൽടൈപ്പ് png ആണെന്ന് ശ്രദ്ധിക്കുക)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # സൃഷ്ടിച്ച ചിത്രത്തെ പിടിക്കുക
        print(generation_response)

        image_url = generation_response.data[0].url # പ്രതികരണത്തിൽ നിന്ന് ചിത്രം URL എടുത്തു കൊണ്ടുവരുക
        generated_image = requests.get(image_url).content  # ചിത്രം ഡൗൺലോഡ് ചെയ്യുക
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ഡിഫോൾട്ട് ഇമേജ് വ്യൂവറിൽ ചിത്രം പ്രദർശിപ്പിക്കുക
        image = Image.open(image_path)
        image.show()

    # ഒഴിവുകൾ പിടിക്കുക
    except openai.BadRequestError as err:
        print(err)

    ```

    # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക

### ചിത്രമുണ്ടാക്കുന്നതിനെക്കുറിച്ചുള്ള കൂടുതൽ വിവരങ്ങൾ

ചിത്രമുണ്ടാക്കുന്ന കോഡ് കൂടുതൽ വിശദമായി നോക്കാം:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
        size='1024x1024',
        n=1
    )
```

- **prompt**, ചിത്രമുണ്ടാക്കാൻ ഉപയോഗിക്കുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റാണ്. ഈ കണക്കിൽ, "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" എന്ന പ്രോംപ്റ്റ് ഉപയോഗിച്ചിരിക്കുന്നു.
- **size**, നിർമ്മിക്കപ്പെടുന്ന ചിത്രത്തിന്റെ വലിപ്പമാണ്. ഈ ഉദാഹരണത്തിൽ, 1024x1024 പിക്‌സലുകൾ ഉള്ള ചിത്രം നിർമ്മിച്ചിട്ടില്ല.
- **n**, നിർമിക്കപ്പെടുന്ന ചിത്രങ്ങളുടെ എണ്ണം. ഈ കണക്കിൽ, രണ്ട് ചിത്രങ്ങൾ നിർമ്മിച്ചിരിക്കുന്നു.

ചിത്രങ്ങൾക്കുമായി നിങ്ങൾ ചെയ്യാവുന്ന മറ്റു കാര്യങ്ങളും അടുത്ത ഭാഗത്ത് ചര്‍ച്ച ചെയ്യാം.

## ചിത്രമുണ്ടാക്കുന്നതിലെ കൂടുതൽ കഴിവുകൾ

ഇതുവരെ, Python-ലെ കുറച്ച് വരികൾ ഉപയോഗിച്ച് ചിത്രമുണ്ടാക്കാൻ സാധിച്ചതിൽ നിങ്ങൾക്ക് ഒരു ധാരണയില്ല. എന്നാൽ, ചിത്രങ്ങൾക്ക് നിങ്ങൾ ചെയ്യാൻ കഴിയുന്ന മറ്റു കാര്യങ്ങൾ ഉണ്ട്.

നിങ്ങൾക്ക് താഴെ കാണുന്നവകളും ചെയ്യാം:

- **എഡിറ്റുകൾ ചെയ്യുക**. നിലവിലുള്ള ഒരു ചിത്രത്തിന് ഒരു മാസ്‌കും ഒരു പ്രോംപ്റ്റും നൽകി ചിത്രം മാറ്റാം. ഉദാഹരണത്തിന്, ഒരു ചിത്രത്തിന്റെ ചില ഭാഗത്ത് എന്തെങ്കിലും ചേർക്കാൻ കഴിയും. നമ്മുടെ റബ്ബിറ്റ് ചിത്രത്തിൽ, ബണ്ണിക്ക് ഒരു ഹാറ്റ് ചേർക്കാം. ഇത് ചെയ്യാൻ, ചിത്രം, മാറ്റം വരുത്തേണ്ട ഭാഗം തിരിച്ചറിയുന്ന മാസ്ക്, കൂടാതെ എന്ത് ചെയ്യണമെന്ന് പറയുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് നൽകുന്നത് ആവശ്യമാണ്.

    ```python
    response = client.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    അടിസ്ഥാന ചിത്രം മട്ടിൽ രാബിറ്റ് മാത്രമുണ്ടാകും, പക്ഷേ ഇതിന് ശേഷം ഹാറ്റ് ബണ്ണിക്കു മുകളിൽ കാണിക്കും.
    
- **വിവിധത്വങ്ങൾ സൃഷ്ടിക്കുക**. നിലവിലുള്ള ഒരു ചിത്രം എടുത്ത് അതിന്റെ വ്യത്യാസങ്ങൾ സൃഷ്ടിക്കാൻ ആവശ്യപ്പെടുന്നതാണ്. ഒരു വകഭേദം സൃഷ്ടിക്കാൻ, ഒരു ചിത്രം, ടെക്സ്റ്റ് പ്രോംപ്റ്റ് ചേർത്ത് താഴെ കാണുന്ന പോലെ കോഡ് നൽകുക:

    ```python
    response = client.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
