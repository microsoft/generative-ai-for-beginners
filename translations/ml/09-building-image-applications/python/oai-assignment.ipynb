# ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കൽ

LLM-കൾക്ക് ടെക്സ്റ്റ് ജനറേഷനിൽ മാത്രമല്ല, ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ സൃഷ്ടിക്കാനും കഴിയും. ചിത്രങ്ങൾ ഒരു മോഡാലിറ്റി ആയി ഉള്ളത് മെഡടെക്, ആർക്കിടെക്ചർ, ടൂറിസം, ഗെയിം ഡെവലപ്പ്മെന്റ് തുടങ്ങിയ നിരവധി മേഖലകളിൽ വളരെ ഉപകാരപ്രദമായിരിക്കാം. ഈ അധ്യായത്തിൽ, ഏറ്റവും ജനപ്രിയമായ രണ്ട് ഇമേജ് ജനറേഷൻ മോഡലുകൾ, DALL-Eയും Midjourneyയും നാം പരിശോധിക്കും.

## DALL-Eയും Midjourneyയും എങ്ങനെ പ്രവർത്തിക്കുന്നു

ആദ്യം, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ഒരു ജനറേറ്റീവ് AI മോഡലാണ്, ട്രാൻസ്ഫോർമർ ആർക്കിടെക്ചർ അടിസ്ഥാനമാക്കിയുള്ള *ഓട്ടോറെഗ്രസീവ് ട്രാൻസ്ഫോർമർ* ഉപയോഗിക്കുന്നു.

*ഓട്ടോറെഗ്രസീവ് ട്രാൻസ്ഫോർമർ* ഒരു മോഡൽ ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ എങ്ങനെ സൃഷ്ടിക്കുന്നുവെന്ന് നിർവചിക്കുന്നു, ഇത് ഓരോ പിക്‌സലും ഒരേസമയം സൃഷ്ടിച്ച്, സൃഷ്ടിച്ച പിക്‌സലുകൾ ഉപയോഗിച്ച് അടുത്ത പിക്‌സൽ സൃഷ്ടിക്കുന്നു. ഒരു ന്യൂറൽ നെറ്റ്‌വർക്കിലെ പല ലെയറുകൾ കടന്നുപോകുന്നു, ചിത്രമാകുന്നത് വരെ.

ഈ പ്രക്രിയയിലൂടെ, DALL-E സൃഷ്ടിക്കുന്ന ചിത്രത്തിലെ ഗുണങ്ങൾ, വസ്തുക്കൾ, സ്വഭാവങ്ങൾ എന്നിവ നിയന്ത്രിക്കുന്നു. എന്നാൽ, DALL-E 2, 3 എന്നിവ സൃഷ്ടിക്കുന്ന ചിത്രത്തെ കൂടുതൽ നിയന്ത്രിക്കാൻ കഴിയും.


## നിങ്ങളുടെ ആദ്യ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കൽ

അപ്പോൾ ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കാൻ എന്ത് വേണം? നിങ്ങൾക്ക് താഴെ പറയുന്ന ലൈബ്രറികൾ ആവശ്യമുണ്ട്:

- **python-dotenv**, നിങ്ങളുടെ രഹസ്യങ്ങൾ കോഡിൽ നിന്ന് വേർതിരിച്ച് *.env* ഫയലിൽ സൂക്ഷിക്കാൻ ഈ ലൈബ്രറി ഉപയോഗിക്കാൻ ശക്തമായി ശിപാർശ ചെയ്യുന്നു.
- **openai**, OpenAI API-യുമായി ഇടപഴകാൻ നിങ്ങൾ ഉപയോഗിക്കുന്ന ലൈബ്രറി.
- **pillow**, Python-ൽ ചിത്രങ്ങളുമായി പ്രവർത്തിക്കാൻ.
- **requests**, HTTP അഭ്യർത്ഥനകൾ നടത്താൻ സഹായിക്കാൻ.

1. താഴെ പറയുന്ന ഉള്ളടക്കത്തോടെ *.env* എന്ന ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```


1. മുകളിൽ പറയുന്ന ലൈബ്രറികൾ *requirements.txt* എന്ന ഫയലിൽ താഴെപറയുന്ന വിധം ശേഖരിക്കുക:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. തുടർന്ന്, വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിച്ച് ലൈബ്രറികൾ ഇൻസ്റ്റാൾ ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows-ന്, നിങ്ങളുടെ വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിക്കുകയും സജീവമാക്കുകയും ചെയ്യാൻ താഴെ കൊടുത്തിരിക്കുന്ന കമാൻഡുകൾ ഉപയോഗിക്കുക:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന ഫയലിൽ താഴെ കൊടുത്തിരിക്കുന്ന കോഡ് ചേർക്കുക:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv

    # dotenv ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()

    # OpenAI ഒബ്ജക്റ്റ് സൃഷ്ടിക്കുക
    client = OpenAI()


    try:
        # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
        generation_response = client.images.generate(
            model="dall-e-3",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
            size='1024x1024',
            n=1
        )
        # സൂക്ഷിച്ചിരിക്കുന്ന ചിത്രത്തിനുള്ള ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')

        # ഡയറക്ടറി ഇല്ലെങ്കിൽ, അത് സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ചിത്രം പാത ആരംഭിക്കുക (ഫയൽടൈപ്പ് png ആയിരിക്കണം എന്ന് ശ്രദ്ധിക്കുക)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # സൃഷ്ടിച്ച ചിത്രം ലഭിക്കുക
        print(generation_response)

        image_url = generation_response.data[0].url # പ്രതികരണത്തിൽ നിന്ന് ചിത്രം URL എടുക്കുക
        generated_image = requests.get(image_url).content  # ചിത്രം ഡൗൺലോഡ് ചെയ്യുക
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ഡിഫോൾട്ട് ഇമേജ് വ്യൂവറിൽ ചിത്രം പ്രദർശിപ്പിക്കുക
        image = Image.open(image_path)
        image.show()

    # исключения പിടിക്കുക
    except client.error.InvalidRequestError as err:
        print(err)

    ```

    import openai

    import dotenv

- അതിനുശേഷം, ഒരു ഒബ്ജക്റ്റ് സൃഷ്ടിക്കുന്നു, ഇത് നിങ്ങളുടെ ``.env`` ഫയലിൽ നിന്നുള്ള API കീ ലഭിക്കും.

        # OpenAI ഒബ്ജക്റ്റ് സൃഷ്ടിക്കുക

    # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
    generation_response = client.images.generate(
        model="dall-e-3",

    മുകളിൽ കൊടുത്തിരിക്കുന്ന കോഡ് ഒരു JSON ഒബ്ജക്റ്റ് ആയി പ്രതികരിക്കുന്നു, അതിൽ സൃഷ്ടിച്ച ചിത്രത്തിന്റെ URL ഉൾപ്പെടുന്നു. ആ URL ഉപയോഗിച്ച് ചിത്രം ഡൗൺലോഡ് ചെയ്ത് ഫയലായി സംരക്ഷിക്കാം.

- അവസാനം, ചിത്രം തുറന്ന് സാധാരണ ഇമേജ് വ്യൂവറിലൂടെ പ്രദർശിപ്പിക്കുന്നു:

    image = Image.open(image_path)

### ചിത്രസൃഷ്ടിയെക്കുറിച്ചുള്ള കൂടുതൽ വിശദാംശങ്ങൾ

ചിത്രം സൃഷ്ടിക്കുന്ന കോഡ് കൂടുതൽ വിശദമായി നോക്കാം:

```python
generation_response = client.images.generate(
        model="dall-e-3",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
        size='1024x1024',
        n=1
    )
```

- **prompt**, ചിത്രം സൃഷ്ടിക്കാൻ ഉപയോഗിക്കുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് ആണ്. ഈ ഉദാഹരണത്തിൽ, "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" എന്ന പ്രോംപ്റ്റ് ഉപയോഗിക്കുന്നു.
- **size**, സൃഷ്ടിക്കുന്ന ചിത്രത്തിന്റെ വലിപ്പം ആണ്. ഈ ഉദാഹരണത്തിൽ, 1024x1024 പിക്‌സൽ വലിപ്പമുള്ള ചിത്രം സൃഷ്ടിക്കുന്നു.
- **n**, സൃഷ്ടിക്കുന്ന ചിത്രങ്ങളുടെ എണ്ണം ആണ്. ഈ ഉദാഹരണത്തിൽ, രണ്ട് ചിത്രങ്ങൾ സൃഷ്ടിക്കുന്നു.

ചിത്രങ്ങളുമായി ചെയ്യാൻ കഴിയുന്ന കൂടുതൽ കാര്യങ്ങൾ അടുത്ത ഭാഗത്ത് വിശദീകരിക്കും.

## ചിത്രസൃഷ്ടിയുടെ അധിക ശേഷികൾ

ഇതുവരെ, Python-ൽ കുറച്ച് വരികളിൽ ഒരു ചിത്രം എങ്ങനെ സൃഷ്ടിക്കാമെന്ന് നിങ്ങൾ കണ്ടു. എന്നാൽ, ചിത്രങ്ങളുമായി ചെയ്യാൻ കഴിയുന്ന കൂടുതൽ കാര്യങ്ങൾ ഉണ്ട്.

നിങ്ങൾക്ക് താഴെ പറയുന്ന കാര്യങ്ങളും ചെയ്യാം:

- **എഡിറ്റുകൾ നടത്തുക**. നിലവിലുള്ള ഒരു ചിത്രത്തിന് മാസ്കും പ്രോംപ്റ്റും നൽകി, ചിത്രം മാറ്റം വരുത്താം. ഉദാഹരണത്തിന്, ഒരു ചിത്രത്തിന്റെ ഒരു ഭാഗത്ത് എന്തെങ്കിലും ചേർക്കാം. നമ്മുടെ ബണ്ണി ചിത്രത്തെക്കുറിച്ച് ചിന്തിക്കൂ, ബണ്ണിക്ക് ഒരു ടോപ്പി ചേർക്കാം. അതിന്, ചിത്രം, മാസ്ക് (മാറ്റം വരുത്തേണ്ട ഭാഗം തിരിച്ചറിയുന്ന) കൂടാതെ എന്ത് ചെയ്യണമെന്ന് പറയുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് നൽകണം.

    ```python
    response = openai.images.edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    അടിസ്ഥാന ചിത്രം വെറും കുതിരയിലിരിക്കുന്ന ബണ്ണി മാത്രമേ ഉള്ളൂ, എന്നാൽ അന്തിമ ചിത്രത്തിൽ ബണ്ണിക്ക് ടോപ്പി ഉണ്ടായിരിക്കും.
    
- **വേരിയേഷനുകൾ സൃഷ്ടിക്കുക**. നിലവിലുള്ള ഒരു ചിത്രം എടുത്ത് അതിന്റെ വ്യത്യസ്ത രൂപങ്ങൾ സൃഷ്ടിക്കണമെന്ന് ആവശ്യപ്പെടാം. ഒരു വേരിയേഷൻ സൃഷ്ടിക്കാൻ, ചിത്രം, ടെക്സ്റ്റ് പ്രോംപ്റ്റ് എന്നിവ നൽകുകയും താഴെ കാണുന്ന പോലെ കോഡ് എഴുതുകയും ചെയ്യാം:

    ```python
    response = openai.images.create_variation(
      image=open("bunny-lollipop.png", "rb"),
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```
