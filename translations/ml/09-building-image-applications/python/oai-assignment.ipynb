# ചിത്രം സൃഷ്ടി അപേക്ഷനങ്ങൾ നിർമ്മിക്കുന്നു (OpenAI)

LLM-കൾക്ക് ടെക്‌സ്‌റ്റ് സൃഷ്ടിക്കാനുപരി കൂടെയുണ്ട്. ടെക്‌സ്‌റ്റ് വിവരണങ്ങളിൽ നിന്നും ചിത്രങ്ങൾ സൃഷ്ടിക്കാം. മെഡടെക്, ശില്പശാല, ടൂറിസം, ഗെയിം വികസനം, മാർക്കറ്റിംഗ് തുടങ്ങിയ മേഖലകളിൽ ചിത്രങ്ങൾ മോഡാലിറ്റി ആയി ഉപയോഗപ്രദമാണ്. ഈ പാഠത്തിൽ നമ്മൾ OpenAI പ്ലാറ്റ്ഫോമിലെ ഇന്നത്തെ **GPT Image** മോഡലുകളുമായി പ്രവർത്തിക്കുന്നു.

## പഠനലക്ഷ്യങ്ങൾ

ഈ പാഠം അവസാനിച്ച് നിങ്ങൾക്ക് സാധ്യമാകുന്നത്:

- ഗ്രാഫിക്കൽ സൃഷ്ടി എന്ന് എന്താണെന്ന്, എവിടെ ഇതു ഉപയോഗപ്പെടുത്താമെന്ന് വിശദീകരിക്കുക.
- `gpt-image` മോഡൽ കുടുംബത്തെ ഇനി ആർത്ഥമാക്കുക, പഴയ DALL·E മോഡലുകളുമായി ഇതിനു ഉള്ള വ്യത്യാസം മനസ്സിലാക്കുക.
- ഒരു ചിത്രം സൃഷ്ടി അപേക്ഷനം നിർമ്മിക്കുകയും ചിത്രങ്ങൾ എഡിറ്റ് ചെയ്യുകയും ചെയ്യുക.

## ചിത്രം സൃഷ്ടി എന്താണ്?

ചിത്രം സൃഷ്ടി മോഡലുകൾ ഒരു ടെക്‌സ്‌റ്റ് പ്രൊംപ്റ്റിൽ നിന്ന് ചിത്രങ്ങൾ ഉണ്ടാക്കുന്നു. `gpt-image` പോലുള്ള ആധുനിക മോഡലുകൾ പരിശീലനത്തിനിടയിൽ ടെക്‌സ്‌റ്റിന്റെയും ചിത്രങ്ങളുടെയും ബന്ധം പഠിക്കുന്നു, പിന്നീട് ക്രമաբար റാൻഡം ശബ്ദത്തെ (noise) നിങ്ങളുടെ പ്രൊംപ്റ്റിനോട് അനുയോജ്യമായ ഒരു ചിത്രമായി മാറ്റുന്നു.

രണ്ട് പ്രശസ്തമായ ചിത്രം മോഡൽ കുടുംബങ്ങൾ:

- **`gpt-image` (OpenAI)** — ഈ പാഠത്തിൽ ഉപയോഗിക്കുന്ന നിലവിലെ തലമുറ. ഇത് ടെക്‌സ്‌റ്റിൽ നിന്നുള്ള ചിത്രസൃഷ്ടി, ചിത്ര എഡിറ്റിംഗ് (മാസ്ക് ഉപയോഗിച്ച് ഇൻപെയിന്റിംഗ്) എന്നിവ പിന്തുണയ്ക്കുന്നു.
- **Midjourney** — സ്വതന്ത്ര സർവീസും Discord അടിസ്ഥാനത്തിലുള്ള പ്രവൃത്തിസിസ്റ്റവും ഉള്ള പ്രശസ്തമായ മൂന്നാം പക്ഷ മോഡൽ.

> പഴയ OpenAI ചിത്രം മോഡലുകൾ — **DALL·E 2** ഒപ്പം **DALL·E 3** — ലെഗസി ആണ്. DALL·E 3 പുതിയ വിതരണംക്കായി ലഭ്യമല്ല, `create_variation` പോലുള്ള ഫീച്ചറുകൾ DALL·E 2-ൽ മാത്രമേ ഉണ്ടായിട്ടുള്ളൂ. പുതിയ അപ്ലിക്കേഷനുകൾക്കായി `gpt-image` മോഡലുകൾ ഉപയോഗിക്കുക.

> **പ്രധാനം:** `gpt-image` മോഡലുകൾ സൃഷ്ടിച്ച ചിത്രം **base64** (`b64_json`) ആയി നൽകുന്നു, URL അല്ല. നിങ്ങളുടെ കോഡ് ബേസ്64 സ്ട്രിംഗ് ഡികോഡ് ചെയ്ത് ബൈറ്റായി മാറ്റുകയും സേവ് ചെയ്യുകയും ചെയ്യുന്നു — ഡൗൺലോഡ് ചെയ്യാനായി ചിത്രം URL ഇല്ല.


## ആദ്യ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ ബിൽഡ് ചെയ്യൽ

ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കാൻ എന്തെല്ലാം ആവശ്യമുള്ളതാണ്? നിങ്ങൾക്ക് താഴെപ്പറയുന്ന ലൈബ്രറികൾ ആവശ്യമുണ്ട്:

- **python-dotenv**, നിങ്ങളുടെ രഹസ്യങ്ങൾ കോഡിന്റെ പുറത്ത് *.env* ഫയലിൽ സൂക്ഷിക്കാൻ ഈ ലൈബ്രറി ഉപയോഗിക്കാൻ വളരെ ശിപാർശ ചെയ്യപ്പെടുന്നു.
- **openai**, OpenAI API-യുമായി ആശയവിനിമയം നടത്താൻ നിങ്ങള് ഇതാണ് ഉപയോഗിക്കുക.
- **pillow**, Python-ൽ ചിത്രങ്ങളുമായി പ്രവർത്തിക്കാൻ.


1. താഴെക്കദേശ内容 ഉള്ള *.env* എന്നൊരു ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    OPENAI_API_KEY='<add your OpenAI key here>'
    ```



1. മുകളിൽ പറയുന്ന ലൈബ്രറികൾ *requirements.txt* എന്ന ഫയലിൽ താഴെ കാണുന്ന പോലെ ശേഖരിക്കുക:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. തുടർന്ന്, ഒരു വിർച്ച്വൽ എൻവയോൺമെന്റ് ഉണ്ടാക്കി ലൈബ്രറികള്‍ ഇൻസ്റ്റാൾ ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> ويندوോസ്‌ക്കായി, നിങ്ങളുടെ വേർച്വൽ എൻവയരൊൺമെന്റ് സൃഷ്ടിക്കുകയും സജീവമാക്കുകയും ചെയ്യാൻ താഴെ കൊടുത്തിരിക്കുന്ന കമാൻഡുകൾ ഉപയോഗിക്കുക:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന പേരിലുള്ള ഫയലിൽ താഴെ കൊടുത്തിരിക്കുന്ന കോഡ് ചേർക്കുക:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai

    # dotenv ഇമ്പോർട്ട് ചെയ്യുക
    dotenv.load_dotenv()

    # OpenAI ഒബ്ജക്റ്റ് സൃഷ്ടിക്കുക (.env-ലുണ്ടായ OPENAI_API_KEY വായിക്കുന്നു)
    client = openai.OpenAI()


    try:
        # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
        generation_response = client.images.generate(
            model="gpt-image-1",
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്‌സ്‌റ്റ് ഇവിടെ നൽകി
            size='1024x1024',
            n=1
        )
        # സൂക്ഷിച്ചിരിക്കുന്ന ചിത്രത്തിന് ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')

        # ഡയറക്ടറിയില്ലെങ്കിൽ, അത് സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ചിത്രം പാത്ത് ആരംഭിക്കുക (ഫയൽടൈപ്പ് png ആകണം ശ്രദ്ധിക്കൂ)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image മോഡലുകൾ ചിത്രം base64 (b64_json) ആകാരം തിരിച്ചു നൽകുന്നു, URL അല്ല
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # ഡീഫോൾട്ട് ചിത്രം വ്യൂവറിൽ ചിത്രം പ്രദർശിപ്പിക്കുക
        image = Image.open(image_path)
        image.show()

    # exceptionകളെ പിടിക്കുക
    except openai.BadRequestError as err:
        print(err)

    ```

ഈ കോഡ് വിശദീകരിക്കാം:

- ആദ്യം, നമുക്ക് ആവശ്യമായ ലൈബ്രറികൾ ഇറക്കുമതി ചെയ്യുന്നു, ഇതിൽ OpenAI ലൈബ്രറി, dotenv ലൈബ്രറി, base64 മോഡ്യൂൾ, Pillow ലൈബ്രറി എന്നിവ ഉൾപ്പെടുന്നു.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    import openai
    ```

- അതിനു ശേഷം, ഞങ്ങൾ ക്ലയന്റ് സൃഷ്ടിക്കുന്നു, ഇത് നിങ്ങളുടെ ``.env``-ൽ നിന്നുള്ള API കീ വായിക്കുന്നു.

    ```python
    # OpenAI ഒബ്ജക്റ്റ് സൃഷ്ടിക്കുക
    client = openai.OpenAI()
    ```

- അടുത്തതായി, ചിത്രമുണ്ടാക്കുന്നു:

    ```python
    generation_response = client.images.generate(
        model="gpt-image-1",
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രാംപ്‌റ്റ്‌ ടെക്സ്റ്റ് ഇവിടെ എഴുതുക
        size='1024x1024',
        n=1
    )
    ```

    `gpt-image` മോഡലുകൾ ചിത്രം **base64** സ്ട്രിംഗായി `data[0].b64_json`-ൽ നൽകുന്നു. അത് ബൈറ്റുകളായി ഡീകോഡ് ചെയ്ത് ഫയലിലേക്ക് എഴുതുന്നു — ഡൗൺലോഡ് ചെയ്യാനുള്ള URL ഇല്ല.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- അവസാനമായി, ചിത്രം തുറന്ന് സാധാരണ ഇമേജ് വിന്യസികൻ ഉപയോഗിച്ച് അത് പ്രദർശിപ്പിക്കുന്നു:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ചിത്ര നിർമ്മാണത്തെ കുറിച്ചുള്ള കൂടുതൽ വിശദാംശങ്ങൾ

`images.generate` എന്നതിന്റെ പാരാമീറ്ററുകൾ പരിശോധിക്കാം:

- **model**, ഉപയോഗിക്കുന്ന ചിത്ര മോഡലിന്റെ പേര് (ഉദാഹരണത്തിന് `gpt-image-1`).
- **prompt**, ചിത്രം സൃഷ്ടിക്കാൻ ഉപയോഗിക്കുന്ന വാചക നിർദ്ദേശം. ഇവിടെ "തേൻഹോലി പിടിച്ച കുതിരയില്ലെയുന്ന തുറരു മിദങ്ങി നിന്ന് ദഫോദില്ലുകൾ വളരുന്ന മൂടൽമഞ്ഞുള്ള കാടാസ് പുള്ളി".
- **size**, നിർമ്മിക്കുന്ന ചിത്രത്തിന്റെ വലിപ്പം (ഉദാഹരണത്തിന് `1024x1024`, `1536x1024`, `1024x1536`, അല്ലെങ്കിൽ `"auto"`).
- **n**, നിർമ്മിക്കുന്ന ചിത്രങ്ങളുടെ എണ്ണം. ഇവിടെ നാം ഒരു ചിത്രം നിർമ്മിക്കുന്നു.

> ചിത്ര മോഡലുകൾ `temperature` പിരുമീറ്റർ സ്വീകരിക്കുന്നില്ല — അത് ടെക്സ്റ്റ് ജനറേഷൻ നിയന്ത്രണമാണ്. വൈവിധ്യം വേണമെങ്കിൽ API വീണ്ടും വിളിക്കൂ; കുറയ്ക്കാൻ, നിങ്ങളുടെ പ്രോണ്പ്റ്റ് കൂടുതൽ വ്യക്തമായതാക്കൂ.

## ചിത്ര നിർമ്മാണത്തിന്റെ അധിക ശേഷികൾ

പൈതന്ന് ചില വരികളിൽ ഒരു ചിത്രം എങ്ങനെയാണ് സൃഷ്ടിക്കാമെന്ന് നിങ്ങൾ കണ്ടു. `gpt-image` മോഡലുകൾ നിലവിലുള്ള ഒരു ചിത്രം **സമ്പാദിപ്പിക്കാനും** കഴിയും. ഒരു ചിത്രം, ഒരു ഓപ്ഷണൽ **മാസ്റ്റ്** (മാറ്റം വരുത്തേണ്ട ഭാഗം അടയാളപ്പെടുത്തുന്നതിന്), കൂടാതെ ഒരു പ്രോണ്പ്റ്റ് നൽകുമ്പോൾ, ചിത്രം的一ഭാഗം മാറ്റാൻ സാധിക്കും — ഉദാഹരണത്തിന്, നമ്മുടെ പുള്ളിക്ക് തൊപ്പി ചേർക്കുക.

```python
response = client.images.edit(
  model="gpt-image-1",
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# എഡിറ്റുകളും ബേസ്64 ആയി തിരിച്ചുകിട്ടുന്നു
```
