# ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കൽ

LLM-കൾക്ക് ടെക്സ്റ്റ് ജനറേഷനിൽ മാത്രമല്ല കൂടുതൽ കഴിവുകൾ ഉണ്ട്. ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ സൃഷ്ടിക്കാനും സാധിക്കും. ചിത്രങ്ങൾ ഒരു മോഡാലിറ്റി ആയി ഉപയോഗിക്കുന്നത് മെഡടെക്, ആർക്കിടെക്ചർ, ടൂറിസം, ഗെയിം ഡെവലപ്പ്മെന്റ് തുടങ്ങിയ നിരവധി മേഖലകളിൽ വളരെ പ്രയോജനകരമായിരിക്കാം. ഈ അധ്യായത്തിൽ, ഏറ്റവും പ്രശസ്തമായ രണ്ട് ഇമേജ് ജനറേഷൻ മോഡലുകൾ, DALL-Eയും Midjourneyയും പരിചയപ്പെടാം.

## പരിചയം

ഈ പാഠത്തിൽ, നാം ചർച്ച ചെയ്യുന്നത്:

- ഇമേജ് ജനറേഷൻ എന്താണെന്നും അത് എങ്ങനെ പ്രയോജനകരമാണെന്നും.
- DALL-Eയും Midjourneyയും എന്താണെന്നും അവ എങ്ങനെ പ്രവർത്തിക്കുന്നുവെന്നും.
- ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ എങ്ങനെ നിർമ്മിക്കാമെന്നും.

## എന്തുകൊണ്ട് ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കണം?

ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷനുകൾ ജനറേറ്റീവ് AI-യുടെ കഴിവുകൾ പരീക്ഷിക്കാൻ മികച്ച മാർഗമാണ്. ഉദാഹരണത്തിന്, ഇവ ഉപയോഗിക്കാം:

- **വിവിധ വ്യവസായങ്ങളിൽ പ്രയോഗം**. മെഡടെക്, ടൂറിസം, ഗെയിം ഡെവലപ്പ്മെന്റ് തുടങ്ങിയ വ്യവസായങ്ങളിൽ ചിത്രങ്ങൾ സൃഷ്ടിക്കാൻ ഇവ ഉപയോഗിക്കാം.

## DALL-Eയും Midjourneyയും എന്താണ്?

## DALL-Eയും Midjourneyയും എങ്ങനെ പ്രവർത്തിക്കുന്നു

ആദ്യം, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ഒരു ജനറേറ്റീവ് AI മോഡലാണ്, ട്രാൻസ്ഫോർമർ ആർക്കിടെക്ചർ അടിസ്ഥാനമാക്കിയുള്ള *ഓട്ടോറെഗ്രസീവ് ട്രാൻസ്ഫോർമർ* ഉപയോഗിക്കുന്നു.

*ഓട്ടോറെഗ്രസീവ് ട്രാൻസ്ഫോർമർ* ഒരു മോഡൽ ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ എങ്ങനെ സൃഷ്ടിക്കുന്നുവെന്ന് നിർവചിക്കുന്നു, ഇത് ഓരോ പിക്‌സലും ഒരിക്കൽ സൃഷ്ടിച്ച്, സൃഷ്ടിച്ച പിക്‌സലുകൾ ഉപയോഗിച്ച് അടുത്ത പിക്‌സൽ സൃഷ്ടിക്കുന്നു. ഒരു ന്യൂറൽ നെറ്റ്‌വർക്കിലെ പല ലെയറുകൾ കടന്നുപോകുന്നു, ചിത്രമാകുന്നത് വരെ.

ഈ പ്രക്രിയയിലൂടെ, DALL-E സൃഷ്ടിക്കുന്ന ചിത്രത്തിലെ ഗുണങ്ങൾ, വസ്തുക്കൾ, സ്വഭാവങ്ങൾ എന്നിവ നിയന്ത്രിക്കുന്നു. എന്നാൽ, DALL-E 2, 3 മോഡലുകൾ സൃഷ്ടിക്കുന്ന ചിത്രത്തെ കൂടുതൽ നിയന്ത്രിക്കാൻ കഴിയും.


## നിങ്ങളുടെ ആദ്യ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കൽ

അപ്പോൾ ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കാൻ എന്ത് വേണം? നിങ്ങൾക്ക് താഴെ പറയുന്ന ലൈബ്രറികൾ ആവശ്യമുണ്ട്:

- **python-dotenv**, നിങ്ങളുടെ രഹസ്യങ്ങൾ കോഡിൽ നിന്ന് വേർതിരിച്ച് *.env* ഫയലിൽ സൂക്ഷിക്കാൻ ഈ ലൈബ്രറി ഉപയോഗിക്കാൻ ശക്തമായി ശിപാർശ ചെയ്യുന്നു.
- **openai**, OpenAI API-യുമായി ഇടപഴകാൻ നിങ്ങൾ ഉപയോഗിക്കുന്ന ലൈബ്രറി.
- **pillow**, Python-ൽ ചിത്രങ്ങളുമായി പ്രവർത്തിക്കാൻ.
- **requests**, HTTP അഭ്യർത്ഥനകൾ നടത്താൻ സഹായിക്കാൻ.

1. താഴെ പറയുന്ന ഉള്ളടക്കത്തോടെ *.env* എന്ന ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    ```

    നിങ്ങളുടെ റിസോഴ്‌സിനുള്ള "Keys and Endpoint" വിഭാഗത്തിൽ Azure പോർട്ടലിൽ ഈ വിവരങ്ങൾ കണ്ടെത്തുക.


1. മുകളിൽ പറയുന്ന ലൈബ്രറികൾ *requirements.txt* എന്ന ഫയലിൽ താഴെപറയുന്ന വിധം ശേഖരിക്കുക:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. തുടർന്ന്, വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിച്ച് ലൈബ്രറികൾ ഇൻസ്റ്റാൾ ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows-ന്, നിങ്ങളുടെ വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിക്കുകയും സജീവമാക്കുകയും ചെയ്യാൻ താഴെ കൊടുത്തിരിക്കുന്ന കമാൻഡുകൾ ഉപയോഗിക്കുക:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന ഫയലിൽ താഴെ കൊടുത്തിരിക്കുന്ന കോഡ് ചേർക്കുക:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    
    # dotenv ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()
    
    # പരിസ്ഥിതി വ്യത്യാസങ്ങളിൽ നിന്ന് എൻഡ്‌പോയിന്റും കീയും നേടുക
    openai.api_base = os.environ['AZURE_OPENAI_ENDPOINT']
    openai.api_key = os.environ['AZURE_OPENAI_API_KEY']     
    
    # API പതിപ്പ് നിയോഗിക്കുക (DALL-E നിലവിൽ 2023-06-01-preview API പതിപ്പിനായി മാത്രമേ പിന്തുണയ്ക്കപ്പെടുകയുള്ളൂ)
    openai.api_version = '2023-06-01-preview'
    openai.api_type = 'azure'
    
    
    try:
        # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
        generation_response = openai.Image.create(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
            size='1024x1024',
            n=2,
            temperature=0,
        )
        # സൂക്ഷിച്ചിരിക്കുന്ന ചിത്രത്തിനുള്ള ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')
    
        # ഡയറക്ടറി ഇല്ലെങ്കിൽ, അത് സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # ചിത്രം പാത ആരംഭിക്കുക (ഫയൽടൈപ്പ് png ആയിരിക്കണം എന്ന് ശ്രദ്ധിക്കുക)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # സൃഷ്ടിച്ച ചിത്രം പുനഃപ്രാപ്തമാക്കുക
        image_url = generation_response["data"][0]["url"]  # പ്രതികരണത്തിൽ നിന്ന് ചിത്രം URL എടുക്കുക
        generated_image = requests.get(image_url).content  # ചിത്രം ഡൗൺലോഡ് ചെയ്യുക
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # ഡിഫോൾട്ട് ചിത്രം കാണിക്കുന്ന പ്രോഗ്രാമിൽ ചിത്രം പ്രദർശിപ്പിക്കുക
        image = Image.open(image_path)
        image.show()
    
    # исключения പിടിക്കുക
    except openai.InvalidRequestError as err:
        print(err)

    ```

ഈ കോഡ് വിശദീകരിക്കാം:

- ആദ്യം, OpenAI ലൈബ്രറി, dotenv ലൈബ്രറി, requests ലൈബ്രറി, Pillow ലൈബ്രറി എന്നിവ ഉൾപ്പെടെ ആവശ്യമായ ലൈബ്രറികൾ ഇറക്കുമതി ചെയ്യുന്നു.

    ```python
    # ഡോട്ട് എൻവൈറൺമെന്റ് ഇറക്കുമതി ചെയ്യുക
    ```

- അതിനുശേഷം, OpenAI API-യുടെ എൻഡ്പോയിന്റ്, കീ, വേർഷൻ, ടൈപ്പ് എന്നിവ സജ്ജമാക്കുന്നു.

    ```python
    # എൻവയോൺമെന്റ് വേരിയബിളുകളിൽ നിന്ന് എൻഡ്‌പോയിന്റും കീയും നേടുക

    openai.api_key = os.environ['AZURE_OPENAI_API_KEY'] 

    # പതിപ്പും തരം ചേർക്കുക, ആസ്യൂർ പ്രത്യേകമായത്
    ```

- തുടർന്ന്, ചിത്രം സൃഷ്ടിക്കുന്നു:

    ```python
    # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
    generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
    ```

### ചിത്രസൃഷ്ടിയെക്കുറിച്ചുള്ള കൂടുതൽ വിശദാംശങ്ങൾ

ചിത്രം സൃഷ്ടിക്കുന്ന കോഡ് കൂടുതൽ വിശദമായി നോക്കാം:

```python
generation_response = openai.Image.create(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
        size='1024x1024',
        n=2,
        temperature=0,
    )
```

- **prompt**, ചിത്രം സൃഷ്ടിക്കാൻ ഉപയോഗിക്കുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് ആണ്. ഈ ഉദാഹരണത്തിൽ, "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" എന്ന പ്രോംപ്റ്റ് ഉപയോഗിക്കുന്നു.
- **size**, സൃഷ്ടിക്കുന്ന ചിത്രത്തിന്റെ വലിപ്പം ആണ്. ഈ ഉദാഹരണത്തിൽ, 1024x1024 പിക്‌സൽ വലിപ്പമുള്ള ചിത്രം സൃഷ്ടിക്കുന്നു.
- **n**, സൃഷ്ടിക്കുന്ന ചിത്രങ്ങളുടെ എണ്ണം ആണ്. ഈ ഉദാഹരണത്തിൽ, രണ്ട് ചിത്രങ്ങൾ സൃഷ്ടിക്കുന്നു.
- **temperature**, ഒരു Generative AI മോഡലിന്റെ ഔട്ട്പുട്ടിന്റെ യാദൃച്ഛികത നിയന്ത്രിക്കുന്ന പാരാമീറ്റർ ആണ്. 0 മുതൽ 1 വരെയുള്ള മൂല്യം ആണ് temperature, 0 ആണെങ്കിൽ ഔട്ട്പുട്ട് നിർണ്ണായകമാണ്, 1 ആണെങ്കിൽ ഔട്ട്പുട്ട് യാദൃച്ഛികമാണ്. ഡീഫോൾട്ട് മൂല്യം 0.7 ആണ്.

ചിത്രങ്ങളുമായി ചെയ്യാൻ കഴിയുന്ന കൂടുതൽ കാര്യങ്ങൾ അടുത്ത സെക്ഷനിൽ കാണാം.

## ചിത്രസൃഷ്ടിയുടെ അധിക ശേഷികൾ

ഇതുവരെ Python-ൽ കുറച്ച് വരികളിൽ ഒരു ചിത്രം സൃഷ്ടിക്കാൻ കഴിഞ്ഞു. എന്നാൽ, ചിത്രങ്ങളുമായി ചെയ്യാൻ കഴിയുന്ന കൂടുതൽ കാര്യങ്ങൾ ഉണ്ട്.

നിങ്ങൾക്ക് താഴെ പറയുന്ന കാര്യങ്ങളും ചെയ്യാം:

- **എഡിറ്റുകൾ നടത്തുക**. നിലവിലുള്ള ഒരു ചിത്രത്തിന് മാസ്കും പ്രോംപ്റ്റും നൽകി ചിത്രം മാറ്റാം. ഉദാഹരണത്തിന്, ഒരു ചിത്രത്തിന്റെ ഒരു ഭാഗത്ത് എന്തെങ്കിലും ചേർക്കാം. നമ്മുടെ ബണ്ണി ചിത്രത്തെക്കുറിച്ച് ചിന്തിക്കൂ, ബണ്ണിക്ക് ഒരു ടോപ്പി ചേർക്കാം. അതിന് ചിത്രം, മാറ്റം വരുത്തേണ്ട ഭാഗം തിരിച്ചറിയുന്ന മാസ്ക്, എന്ത് ചെയ്യണമെന്ന് പറയുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് എന്നിവ നൽകണം.

    ```python
    response = openai.Image.create_edit(
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response['data'][0]['url']
    ```

    അടിസ്ഥാന ചിത്രം ബണ്ണിയെ മാത്രമേ ഉൾക്കൊള്ളൂ, എന്നാൽ അന്തിമ ചിത്രത്തിൽ ബണ്ണിക്ക് ടോപ്പി ഉണ്ടാകും.
