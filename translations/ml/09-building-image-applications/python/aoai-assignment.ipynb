# ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷനുകൾ നിർമ്മിക്കുക (അസ്യൂർ ഓപ്പൺഎ.ഐ)

LLMs ടെക്സ്റ്റ് ജനറേഷനിലേക്കു മാത്രമല്ല. നിങ്ങൾക്ക് ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ഇമേജുകൾ സൃഷ്ടിക്കാനുമാകും. ചിത്രങ്ങൾ ഒരു മോഡാലിറ്റിയായിരിക്കുമ്പോൾ മെഡ്ടेक്നോളജി, ആർക്കിടെക്ചർ, ടൂറിസം, ഗെയിം ഡെവലപ്‌മെന്റ്, മാർക്കറ്റിംഗ് തുടങ്ങിയ വിവിധ മേഖലകളിൽ ഇത് പ്രയോജനപ്പെടുന്നു. ഈ പാഠത്തിൽ നാം മൈക്രോസോഫ്റ്റ് ഫൗണ്ടറിയിലെ ഇന്നത്തെ **GPT Image** മോഡലുകളുമായി പ്രവർത്തിക്കുകയാണ്.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠം അവസാനിക്കുന്നത് വരെ നിങ്ങളാകും:

- ഇമേജ് ജനറേഷൻ എന്താണെന്ന് എവിടെ പ്രയോജനപ്പെടുന്നുവെന്ന് വിശദീകരിക്കുക.
- `gpt-image` മോഡൽ കുടുംബത്തെ മനസിലാക്കി, പഴയ DALL·E മോഡലുകളുമായി ഉള്ള വ്യത്യാസം അറിയുക.
- ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിച്ച് ചിത്രങ്ങൾ എഡിറ്റ് ചെയ്യുക.

## ഇമേജ് ജനറേഷൻ എന്നത് എന്താണ്?

ഇമേജ് ജനറേഷൻ മോഡലുകൾ ടെക്സ്റ്റ് പ്രോംപ്റ്റിൽ നിന്നു ചിത്രങ്ങൾ സൃഷ്ടിക്കുന്നു. `gpt-image` പോലെയുള്ള ആധുനിക മോഡലുകൾ പരിശീലന സമയത്ത് ടെക്സ്റ്റും ചിത്രവും തമ്മിലുള്ള ബന്ധം പഠിച്ച്, പിന്നീടു നിരന്തരം അനിയോജ്യമായ ശബ്ദത്തെ ഉണ്ടായതിലും നിങ്ങളുടെ പ്രോംപ്റ്റുമായി പൊരുത്തപ്പെടുന്നതുമായ ചിത്രം ആക്കുന്നു.

രണ്ടെണ്ണം പ്രശസ്തമായ ഇമേജ് മോഡൽ കുടുംബങ്ങൾ:

- **`gpt-image` (ഓപ്പൺഎ.ഐ)** — ഈ പാഠത്തിൽ ഉപയോഗിക്കുന്ന നിലവിലെ തലമുറ. ഇത് ടെക്സ്റ്റ്-টু-ഇമേജ് ജനറേഷനും ചിത്രം എഡിറ്റിംഗിനും (മാസ്‌കോടുകൂടിയ ഇൻപെയിന്റിങ്) പിന്തുണയുണ്ട്.
- **Midjourney** — സ്വന്തം സേവനവും ഡിസ്കോർഡ് അടിസ്ഥാനത്തിൽ പ്രവർത്തിക്കുന്ന പ്രവൃത്തിപദ്ധതിയും ഉള്ള പ്രശസ്ത തൃतीय പക്ഷ മോഡൽ.

> പഴയ ഓപ്പൺഎ.ഐ ഇമേജ് മോഡലുകൾ — **DALL·E 2**നും **DALL·E 3**നും — പാരമ്പര്യമാണ്. DALL·E 3 പുതിയ ഡിപ്പ്ലോയ്മെന്റുകൾക്ക് ലഭ്യമല്ല, `create_variation` പോലുള്ള ഫീച്ചറുകൾ മാത്രമേ DALL·E 2-ൽ ഉണ്ടായിരുന്നുള്ളു. പുതിയ അപ്ലിക്കേഷനുകൾക്കായി `gpt-image` മോഡലുകൾ ഉപയോഗിക്കുക.

മൈക്രോസോഫ്റ്റ് ഫൗണ്ടറിയിൽ, **`gpt-image-2`** ഏറ്റവും പുതിയതും ശേഷിവെള്ളമുള്ളതുമായ ഇമേജ് മോഡലാണ്, ശുപാർശചെയ്യപ്പെട്ട ഡിഫോൾട് മോഡൽ. `gpt-image-1.5`നും `gpt-image-1-mini`നും സാധാരണ ലഭ്യമാണ്.

> **പ്രധാനമായത്:** `gpt-image` മോഡലുകൾ സൃഷ്ടിച്ച ചിത്രത്തെ **base64** (`b64_json`) എന്ന നിലയിൽ നൽകുന്നു, URL ആയി അല്ല. നിങ്ങളുടെ കോഡ് ഈ base64 സ്ട്രിംഗ് ബൈറ്റുകളായി ഡീകോഡ് ചെയ്ത് സേവ് ചെയ്യുന്നു — ഡൗൺലോഡ് ചെയ്യുവാൻ ചിത്രം URL ഇല്ല.


## നിങ്ങളുടെ ആദ്യ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കുക

അതെ, ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കാൻ എന്താണ് ആവശ്യം? നിങ്ങൾക്കു താഴെ കാണുന്ന ലൈബ്രറികൾ വേണം:

- **python-dotenv**, നിങ്ങളുടെ രഹസ്യങ്ങൾ കോഡ് നിന്നും വേർപെട്ട *.env* ഫയലിൽ സൂക്ഷിക്കാന്‍ ഈ ലൈബ്രറി ഉപയോഗിക്കാൻ വളരെ ശിപാർശ ചെയ്യപ്പെടുന്നു.
- **openai**, OpenAI API-യെ സമ്പർക്കിക്കാൻ നിങ്ങൾ ഉപയോഗിക്കുന്ന ലൈബ്രറിയാണ് ഇത്.
- **pillow**, Python-ൽ ചിത്രങ്ങളുമായി പ്രവർത്തിക്കാൻ.

ഇത് കഴിഞ്ഞിട്ടില്ലെങ്കിൽ, [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) പേജിലുള്ള നിർദ്ദേശങ്ങൾ അനുസരിച്ച് ഒരു Microsoft Foundry റിസോഴ്സ് ಮತ್ತು മോഡൽ സൃഷ്ടിക്കുക. മോഡലായി **gpt-image-2** തിരഞ്ഞെടുക്കുക (ഇതാണ് ഏറ്റവും പുതിയ Azure OpenAI ഇമേജ് മോഡൽ; DALL·E പഴയ മോഡലാണ്).

1. താഴെ കാണുന്ന ഉള്ളടക്കത്തോടെ *.env* എന്ന ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    നിങ്ങളുടെ റിസോഴ്‌സിനുള്ള Microsoft Foundry പോർട്ടലിലെ "Deployments" വിഭാഗത്തിൽ ഈ വിവരങ്ങൾ കണ്ടെത്തുക.



1. മുകളിൽ പറയുന്ന ലൈബ്രറിയുകൾ *requirements.txt* എന്ന പേരിലുള്ള ഒരു ഫയലിൽ ശേഖരിക്കുക, ഇതുപോലെ:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. അടുത്തതായി, ഒരു വെർച്വൽ എൻവയോൺമെന്റ് സൃഷ്ടിച്ച് ലൈബ്രറികൾ ഇൻസ്റ്റാൾ ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Windows ൽ, നിങ്ങളുടെ വെർച്വൽ പരിസരം സൃഷ്ടിച്ച് സജീവമാക്കാൻ താഴെ കൊടുത്തിരിക്കുന്ന കമാൻഡുകൾ ഉപയോഗിക്കൂ:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന പേരിലുള്ള ഫയലിൽ താഴെ കൊടുത്തിരിക്കുന്ന കോഡ് ചേർക്കുക:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # dotenv ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()

    # Azure OpenAI (Microsoft Foundry) ക്ലയന്റ് ക്രമീകരിക്കുക.
    # ചിത്ര മോഡലിന് പുതിയ API പതിപ്പ് ആവശ്യം — നിങ്ങളുടെ മോഡലിന് ആവശ്യമായ പതിപ്പ് കാണാൻ Foundry ഡോക്യുമെന്റേഷൻ പരിശോധിക്കുക.
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # ചിത്രം സൃഷ്ടിക്കാൻ ചിത്രം ജനറേഷൻ API ഉപയോഗിക്കുക
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # ഉദാഹരണത്തിന് "gpt-image-2"
        )
        # സംഭരിച്ചിരിക്കുന്ന ചിത്രത്തിന് ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')

        # ഡയറക്ടറി ഇല്ലെങ്കിൽ, അത് സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # ചിത്രം പാത്ത് ആരംഭിക്കുക (ഫയൽടൈപ്പ് png ആയിരിക്കും ശ്രദ്ധിക്കുക)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # gpt-image മോഡലുകൾ ചിത്രം URL അല്ല, base64 (b64_json) ഫോർമാറ്റിൽ തിരികെ നൽകും
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # എന്ന് സാധാരണ ചിത്രം കാണിക്കുന്ന viewer-ൽ ചിത്രം കാണിക്കുക
        image = Image.open(image_path)
        image.show()

    # വ്യത്യാസങ്ങൾ പിടികൂടുക
    except BadRequestError as err:
        print(err)

    ```

ഈ കോഡ് വിശദീകരിക്കാം:

- ആദ്യം, നമുക്ക് ആവശ്യമുള്ള ലൈബ്രറികൾ ഇറക്കുമതി ചെയ്യുന്നു, അതിൽ OpenAI ലൈബ്രറി, dotenv ലൈബ്രറി, base64 മോഡ്യൂൾ, Pillow ലൈബ്രറി എന്നിവ ഉൾപ്പെടുന്നു.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- തുടർന്ന്, *env* ഫയലിൽ നിന്നും പരിസര വേരിയബിളുകൾ ലോഡ് ചെയ്യുന്നു.

    ```python
    # ഡോട്ട്‌എൻവി ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()
    ```

- പിറ്റേന്ന്, Azure OpenAI (Microsoft Foundry) ക്ലയന്റിനെ കോൺഫിഗർ ചെയ്യുന്നു.

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- തുടർന്ന്, ചിത്രം സൃഷ്ടിക്കുന്നു:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # ഇവിടെ നിങ്ങളുടെ പ്രോംപ്‌റ്റ് എഴുതി നൽക്കുക
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    `gpt-image` മോഡലുകൾ ചിത്രം **base64** സ്ട്രിങായി `data[0].b64_json` ൽ നൽകുന്നു. നമുക്ക് അത് ബൈറ്റുകളായി ഡികോഡ് ചെയ്ത് ഫയലിൽ എഴുതി வைക്കാം — ഡൗൺലോഡിനുള്ള URL ഇല്ല.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- അവസാനമായി, ചിത്രം തുറന്ന് സാധാരണ ഇമേജ് വിഉവറിൽ പ്രదర్శിപ്പിക്കുന്നു:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### ചിത്രം സൃഷ്ടിക്കുന്നതിലെ കൂടുതൽ വിശദാംശങ്ങൾ

`images.generate` ന്റെ പാരാമീറ്ററുകൾ നോക്കാം:

- **prompt**, ചിത്രം സൃഷ്ടിക്കാൻ ഉപയോഗിക്കുന്ന ടെക്സ്റ്റ് പ്രാമ്പ്റ്റ്. ഇതിന്റെ ഉദാഹരണം "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, സൃഷ്ടിക്കപ്പെടുന്ന ചിത്രത്തിന്റെ വലുപ്പം (ഉദാ: `1024x1024`, `1536x1024`, `1024x1536`, അല്ലെങ്കിൽ `"auto"`).
- **n**, സൃഷ്ടിക്കപ്പെടുന്ന ചിത്രങ്ങളുടെ എണ്ണം. ഇവിടെ ഒന്ന് സൃഷ്ടിക്കുന്നു.
- **model**, നിങ്ങളുടെ ചിത്രം മോഡൽ ഡിപ്ലോയ്മെന്റിന്റെ പേര് (ഉദാ: `gpt-image-2`).

> ചിത്രം മോഡലുകൾ `temperature` പാരാമീറ്റർ ഉപയോഗിക്കുന്നില്ല — അത് ടെക്സ്റ്റ്-ജനറേഷൻ നിയന്ത്രണമാണ്. വൈവിധ്യം നേടാൻ API വീണ്ടും വിളിക്കൂ; വൈവിധ്യം കുറക്കാൻ നിങ്ങളുടെ പ്രാമ്പ്റ്റ് കൂടുതൽ നിർദിഷ്ടമാക്കൂ.

## ചിത്രസൃഷ്ടിയുടെ അധിക കഴിവുകൾ

പൈതൺ കുറച്ച് വരികളിൽ ചിത്രമെങ്ങാനും നിങ്ങൾ കണ്ടിട്ടുണ്ട്. `gpt-image` മോഡലുകൾ നിലവിലുള്ള ചിത്രവും **എഡിറ്റ്** ചെയ്യാനാകും. ഒരു ചിത്രം, ঐച്ഛിക **മാസ്ക്** (മാറ്റേണ്ട പ്രദേശം അടയാളപ്പെടുത്തുന്നതിന്), ഒരു പ്രാമ്പ്റ്റ് എന്നിവ നൽകി, ചിത്രത്തിന്റെ ഭാഗം മാറ്റാൻ കഴിയും — ഉദാഹരണത്തിന്, ഞങ്ങളുടെ കുഞ്ഞ് ബണ്ണിയ്ക്ക് ഒരു ടോപ്പി ചേർക്കാം.

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# എഡിറ്റുകൾ ബേസ64 ആകൃതിയിലും മടങ്ങി നൽകപ്പെടുന്നു
edited_image = base64.b64decode(response.data[0].b64_json)
```

അടിസ്ഥാന ചിത്രം യശുദ്ധമായ خرگوش മാത്രമാണുള്ളത്; അവസാന ചിത്രം خرگوشയ്ക്ക് ടോപ്പി ചേർത്തതാണ്.
