# ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കുന്നു

ടെക്സ്റ്റ് ജനറേഷനേക്കാൾ കൂടുതൽ കാര്യങ്ങൾക്ക് LLM-കൾ ഉണ്ട്. ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ സൃഷ്‌ടിക്കാനും സാധ്യമാണ്. മെഡ്ടെക്, വാസ്തുവിദ്യ, ടൂറിസം, ഗെയിം ഡെവലപ്പ്മെന്റ് മുതലായ പല മേഖകളിലും ഒരു മോഡാലിറ്റി എന്ന നിലയിൽ ചിത്രങ്ങൾ വളരെ ഉപകാരപ്രദമാണ്. ഈ అధ్యായത്തിൽ, നാം ഏറ്റവും ജനപ്രിയമായ രണ്ട് ഇമേജ് ജനറേഷൻ മോഡലുകൾ, DALL-E-യും Midjourney-യും പരിചയപ്പെടും.

## പരിചയം

ഈ പാഠത്തിൽ, നാം പുകഴ്ചകൾക്ക് സമീപിക്കും:

- ഇമേജ് ജനറേഷൻ എന്താണ്, അത് എന്തുകൊണ്ട് പ്രയോജനകാരിയാണ്.
- DALL-E-യും Midjourney-യും എന്താണ്, ഇവ എങ്ങനെ പ്രവർത്തിക്കുന്നു.
- ഒരു ഇമേജ് ജനറേഷൻ ആപ്പിനെ എങ്ങനെ നിർമ്മിക്കാമെന്ന്.

## പഠന ലക്ഷ്യങ്ങൾ

ഈ പാഠം പൂർത്തിയാക്കിയശേഷം, നിങ്ങൾക്ക് കഴിയും:

- ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കുക.
- മെറ്റാപ്രോംസുകളിലൂടെ നിങ്ങളുടെ അപ്ലിക്കേഷന്റെ പരിധികൾ നിർവ്വചിക്കുക.
- DALL-E-യും Midjourney-യും ഉപയോഗിച്ച് പ്രവർത്തിക്കുക.

## എന്തുകൊണ്ട് ഒരു ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷൻ നിർമ്മിക്കണം?

ജനനാത്മക AI യുടെ കഴിവുകൾ കണ്ടെത്താൻ ഇമേജ് ജനറേഷൻ അപ്ലിക്കേഷനുകൾ മികച്ച മാർഗമാണ്. ഉദാഹരണത്തിന് ഉപയോഗിക്കാവുന്നതാണ്:

- **ചിത്ര എഡിറ്റിംഗ്‌വും സിന്തീസിസും**. ചിത്ര എഡിറ്റിംഗിനും ചിത്ര സൃഷ്ടിക്കും പല തരത്തിലുള്ള ഉപയോഗങ്ങൾക്കായി നിങ്ങൾക്ക് ചിത്രങ്ങൾ സൃഷ്ടിക്കാം.

- **വിവിധ വ്യവസായങ്ങളിലേക്ക് പ്രയോഗം**. മെഡ്ടെക്, ടൂറിസം, ഗെയിം ഡെവലപ്മെന്റ് തുടങ്ങിയ വ്യവസായങ്ങൾക്ക് ചിത്രങ്ങൾ സൃഷ്ടിക്കാനായി ഇവ ഉപയോഗിക്കാം.

## DALL-E හා Midjourney എന്താണ്?

[DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) ഉം [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) ഉം ഏറ്റവും ജനപ്രിയമായ രണ്ട۠ു ഇമേജ് ജനറേഷൻ മോഡലുകളാണ്, നിങ്ങൾക്ക് പ്രോംപ്രുകൾ ഉപയോഗിച്ച് ചിത്രങ്ങൾ സൃഷ്ടിക്കാൻ സഹായിക്കുന്നു.

### DALL-E

DALL-E മുതൽക്കാരാണ്, ഒരു ജെനറേറ്റീവ് AI മോഡൽ, ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ സൃഷ്ടിക്കുന്നു.

> [DALL-E എന്നത് രണ്ട് മോഡലുകളായ CLIP-ഉം ഡിഫ്യൂസ്ഡ് അറ്റൻഷൻ-ഉം ചേർന്നതാണ്](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).

- **CLIP**, ചിത്രങ്ങൾക്കും ടെക്സ്റ്റിനും നിന്നുള്ള ഡാറ്റയുടെ സംഖ്യാത്മക പ്രതിനിധാനങ്ങളായ എമ്പെഡിങ്ങുകൾ സൃഷ്ടിക്കുന്ന മോഡലാണ്.

- **ഡിഫ്യൂസ്ഡ് അറ്റൻഷൻ**, എമ്പെഡിങ്ങുകളിൽ നിന്നുള്ള ചിത്രങ്ങൾ സൃഷ്ടിക്കുന്ന മോഡലാണ്. DALL-Eയെ ചിത്രങ്ങൾക്കും ടെക്സ്റ്റിനും ചേർന്ന് ഉള്ള ഡാറ്റാസെറ്റിൽ പരിശീലിപ്പിച്ചിരിക്കുന്നു, ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്ന് ചിത്രങ്ങൾ സൃഷ്ടിക്കാൻ ഇതുപയോഗിക്കാം. ഉദാഹരണത്തിന്, DALL-E ഉപയോഗിച്ച് ഒരു തലയൊട്ടിയുള്ള പൂച്ചയുടെ ചിത്രമോ, മോഹോക്കുള്ള നായയുടെ ചിത്രമോ സൃഷ്ടിക്കാം.

## DALL-Eയും Midjourneyയും എങ്ങനെ പ്രവർത്തിക്കുന്നു

ആദ്യം, [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). DALL-E ഒരു ജെനറേറ്റീവ് AI മോഡലാണ്, അത് ട്രാൻസ്ഫോർമർ ആർക്കിടെക്ചറിനോട് അടിസ്ഥാനത്തിലാണ്, *ഓട്ടോറഗ്രസ്സിവ് ട്രാൻസ്ഫോർമർ* ആണ്.

ഒരു *ഓട്ടോറഗ്രസ്സിവ് ട്രാൻസ്ഫോർമർ* മോഡൽ ടെക്സ്റ്റ് വിവരണങ്ങളിൽ നിന്നുള്ള ചിത്രങ്ങൾ എങ്ങനെ സൃഷ്ടിക്കുന്നു എന്ന് നിർവ്വചിക്കുന്നു, അത് ഓരോ പിക്സലും ഒറ്റയ്ക്ക് സൃഷ്ടിച്ച്, പിന്നീട് നിലവിലുള്ള പിക്സലുകൾ ഉപയോഗിച്ച് അടുത്ത പിക്സൽ സൃഷ്ടിക്കുന്നു. ന്യുറൽ നെറ്റ്വർക്കിലെ നിരവധി ലെയറുകളിലൂടെ കടന്നുപോകുന്നു, hasta ചിത്രം പൂർത്തിയാക്കുന്നത് വരെ.

ഈ പ്രക്രിയയിലൂടെ, DALL-E സൃഷ്ടിക്കുന്ന ചിത്രത്തിലെ ഗുണങ്ങൾ, വസ്തുക്കൾ, സ്വഭാവങ്ങൾ മുതലായവ നിയന്ത്രിക്കുന്നു. എന്നാൽ DALL-E 2 ഉം 3 ഉം സൃഷ്ടിക്കുന്ന ചിത്രത്തെ കുറിച്ച് കൂടുതൽ നിയന്ത്രണവും വാഗ്ദാനം ചെയ്യുന്നു,


## നിങ്ങളുടെ ആദ്യ ഇമേജ് സൃഷ്ടി അപ്ലിക്കേഷൻ നിർമ്മിക്കുക

എങ്ങനെ ഒരു ഇമേജ് സൃഷ്ടി അപ്ലിക്കേഷൻ നിർമിക്കാം? നിങ്ങൾക്ക് താഴെയുള്ള ലൈബ്രറികൾ ആവശ്യമാകും:

- **python-dotenv**, നിങ്ങളുടെ രഹസ്യങ്ങൾ പ്രവർത്തന കോഡിൽ നിന്നും വേർതിരിച്ച് *.env* ഫയലിൽ സൂക്ഷിക്കാൻ ഈ ലൈബ്രറി ഉപയോഗിക്കൽ നിർദ്ദേശിക്കുന്നു.
- **openai**, ഈ ലൈബ്രറി OpenAI API-യുമായി ബന്ധപ്പെടാൻ ഉപയോഗിക്കും.
- **pillow**, പൈത്തൺഇൻറെ ചിത്രങ്ങൾ കൈകാര്യം ചെയ്യുന്ന ലൈബ്രറി.
- **requests**, HTTP അഭ്യർത്ഥനകൾ നടത്താൻ സഹായിക്കും.

ഇതുവരെ ചെയ്തിട്ടില്ലെങ്കിൽ, Azure OpenAI റസോഴ്സ് ಮತ್ತು മോഡൽ സൃഷ്ടിക്കാൻ [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) പേജിലെ നിർദ്ദേശങ്ങൾ അനുസരിക്കുക. മോഡലായി **gpt-image-1** തിരഞ്ഞെടുക്കുക (ഇത് നിലവിലെ തലമുറ Azure OpenAI ഇമേജ് മോഡലാണ്; DALL-E 3 പാന്ധ്യംവേദിയ്ക്കപ്പെടുകയും പുതിയ വിനിയോഗങ്ങൾക്ക് ഇനി ലഭ്യമല്ല).

1. താഴെയുള്ള ഉള്ളടക്കത്തോടെ *.env* ഫയൽ സൃഷ്ടിക്കുക:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-1"
    ```

    നിങ്ങളുടെ റസോഴ്‌സിനായി "Deployments" വിഭാഗത്തിൽ Microsoft Foundry പോർട്ടലിൽ ഈ വിവരങ്ങൾ കണ്ടെത്താം.



1. മുകളില്‍ കൊടുത്ത ലൈബ്രറികള്‍ *requirements.txt* എന്ന പേരില്‍ ഫയലില്‍ ഇങ്ങനെ ശേഖരിക്കുക:

    ```text
    python-dotenv
    openai
    pillow
    requests
    ```


1. അടുത്തതായി, ഒരു വിര്‍ച്വല്‍ എന്‍വയോണ്‍മെന്റ് ഉണ്ടാക്കി ലൈബ്രറികള്‍ ഇന്‍സ്റ്റാള്‍ ചെയ്യുക:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow requests
`````

> [!NOTE]
> Windows ല്, നിങ്ങളുടെ വിർച്വൽ എൻവയറൻമന്റ് സൃഷ്ടിക്കുകയും സജീവമാക്കുകയും ചെയ്യാൻ താഴെ കാണുന്ന കമാൻഡുകൾ ഉപയോഗിക്കുക:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. *app.py* എന്ന പേര് ഉള്ള ഫൈലിൽ താഴെ കാണുന്ന കോഡ് ചേർക്കുക:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # ഡോട്ട്എൻവിക്ക് ഇറക്കുമതി ചെയ്യുക
    dotenv.load_dotenv()

    # ആസ്യൂർ ഓപ്പൺഎഐ സേവന ക്ലയന്റ് ക്രമീകരിക്കുക
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # ഇമേജ് ജനറേഷൻ എപി.ഐ ഉപയോഗിച്ച് പ്രതമാറ്റം സൃഷ്ടിക്കുക
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോംപ്റ്റ് ടെക്സ്റ്റ് ഇവിടെ നൽകുക
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # സഞ്ചയിച്ചിരിക്കുന്ന ചിത്രത്തിന്റെ ഡയറക്ടറി സജ്ജമാക്കുക
        image_dir = os.path.join(os.curdir, 'images')
    
        # ഡയറക്ടറി ഇല്ലെങ്കിൽ, അത് സൃഷ്ടിക്കുക
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # ചിത്രം പാത ആരംഭിക്കുക (ഫയൽടൈപ്പ് png ആകണം എന്നതിന്റെ ശ്രദ്ധിക്കുക)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # നിർമ്മിച്ച ചിത്രം പ്രാപിക്കുക
        image_url = generation_response.data[0].url  # പ്രതികരണത്തിൽ നിന്ന് ചിത്രം URL തിരിച്ചെടുക്കുക
        generated_image = requests.get(image_url).content  # ചിത്രം ഡൗൺലോഡ് ചെയ്യുക
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # ഡിഫോൾട്ട് ഇമേജ് വ്യൂവറിൽ ചിത്രം പ്രദർശിപ്പിക്കുക
        image = Image.open(image_path)
        image.show()
    
    # исключения പിടിക്കുക
    except openai.BadRequestError as err:
        print(err)

    ```

    ```python

    ```

- പിന്നീട്, *.env* ഫയലിൽ നിന്നുള്ള എൻവയറൻമെന്റ് വേരിയബിളുകൾ ലോഡ് ചെയ്യുന്നു.

    ```python
    # ഡോട്ട്‌എൻ‌വി ഇറക്കുമതി ചെയ്യുക
    ```

- ശേഷം, Azure OpenAI സർവീസ് ക്ലയന്റ് കോൺഫിഗർ ചെയ്യുന്നു.

    ```python
    # Azure OpenAI സർവീസ് ക്ലയന്റ് ക്രമീകരിക്കുക
    ```

- പിന്നീട്, ആ ചിത്രം ജനറേറ്റ് ചെയ്യുന്നു:

    ```python
    # ഇമേജ് ജനറേഷൻ API ഉപയോഗിച്ച് ഒരു ചിത്രം സൃഷ്ടിക്കുക
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രൊംപ്റ്റ് എഴുതി ഇവിടെ നൽകുക
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    മുകളിൽ കൊടുത്ത കോഡ് ഒരു JSON ഒബ്ജക്റ്റ് ആയി പ്രതികരിക്കുന്നു, അതിൽ ജനറേറ്റ് ചെയ്ത ചിത്രം ഉള്ള URL ഉണ്ട്. ആ URL ഉപയോഗിച്ച് ചിത്രം ഡൗൺലോഡ് ചെയ്ത് ഫയലായി സംരക്ഷിക്കാം.

- അവസാനമായി, ചിത്രം തുറന്ന് സാധാരണ ചിത്രം കാണിക്കുന്ന അപ്പ്ലിക്കേഷൻ ഉപയോഗിച്ച് അത് പ്രദർശിപ്പിക്കുന്നു:

### ചിത്രം ജനറേറ്റ് ചെയ്യുന്നതിന് കൂടുതൽ വിശദാംശങ്ങൾ

ചിത്രം ജനറേറ്റ് ചെയ്യുന്നത് സംബന്ധിച്ച് കോഡ് കൂടുതൽ വിശദമായി നോക്കാം:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # നിങ്ങളുടെ പ്രോമ്പ്റ്റ് മെസ്സേജ് ഇവിടെ നൽകുക
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, ചിത്രം സൃഷ്‌ടിക്കാൻ ഉപയോഗിക്കുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റ് ആണ്. ഈ സാഹചര്യത്തിൽ, നാം "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils" എന്ന പ്രോംപ്റ്റ് ഉപയോഗിക്കുന്നു.
- **size**, ചിത്രം ഏത് വലിപ്പത്തിൽ (പിക്ക്സലുകളിൽ) ഉണ്ടാക്കണമെന്ന് വ്യക്തമാക്കുന്നു. ഈ ഉദാഹരണത്തിൽ, നാം 1024x1024 പിക്‌സൽ വലുപ്പം ജനറേറ്റ് ചെയ്യുകയാണ്.
- **n**, ജനറേറ്റ് ചെയ്യേണ്ട ചിത്രങ്ങളുടെ എണ്ണം സൂചിപ്പിക്കുന്നു. ഈ സാഹചര്യത്തിൽ, ഒരു ചിത്രം മാത്രമാണ് ജനറേറ്റ് ചെയ്യുന്നത്.

ചിത്രങ്ങളുമായി ബന്ധപ്പെട്ട് നാം വരാനിരിക്കുന്ന സെക്‌ഷനിൽ കൂടുതൽ കാര്യങ്ങൾ നോക്കാം.

## ചിത്രജനനത്തിൻറെ അധിക സർവ്വീസുകൾ

ഇതുവരെ നാം കാണുന്നത് പോത്തുംാളിലേക്ക് കുറച്ച് പൈത്തൺ കോഡ് ഉപയോഗിച്ച് ഒരു ചിത്രം എങ്ങനെയാണ് സൃഷ്ടിക്കാം എന്നതാണ്. എന്നാൽ ചിത്രങ്ങളുമായി നിങ്ങൾക്ക് ചെയ്യാൻ കഴിയുന്ന കൂടുതൽ കാര്യങ്ങൾ ഉണ്ട്.

നിങ്ങൾക്ക് താഴെ പറയുന്നവയും ചെയ്യാൻ കഴിയും:

- **തിരുത്തലുകൾ നടത്തുക**. നിലവിലുള്ള ഒരു ചിത്രത്തോടൊപ്പം മാസ്കും ഒരു പ്രോംപ്റ്റും നൽകുകയാണെങ്കിൽ, ചിത്രം മാറ്റം വരുത്താം. ഉദാഹരണത്തിന്, ഒരുഭാഗം ചിത്രത്തിലേക്ക് എന്തെങ്കിലും ചേർക്കാം. ഞങ്ങളുടെ ബണ്ണി ചിത്രത്തെ കരുതിയാൽ, ബണ്ണിക്ക് മേൽ ഒരു ടോപ്പി ചേർക്കാം. അതിനായി ചിത്രവും, മാറ്റം വരുത്തേണ്ട ഭാഗം നിർദേശിക്കുന്ന മാസ്കും, എന്ത് ചെയ്യണമെന്നടക്കം പറയുന്ന ടെക്സ്റ്റ് പ്രോംപ്റ്റും നൽകണം.

    ```python
    response = client.images.edit(
      model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
      image=open("base_image.png", "rb"),
      mask=open("mask.png", "rb"),
      prompt="An image of a rabbit with a hat on its head.",
      n=1,
      size="1024x1024"
    )
    image_url = response.data[0].url
    ```

    അടിസ്ഥാന ചിത്രം മാത്രമേ ബണ്ണി അടങ്ങിയുള്ളൂ ഉള്ളൂ, പക്ഷേ അവസാന ചിത്രത്തിൽ ബണ്ണിക്ക് മേൽ ടോപ്പി ഉണ്ടാകും.
