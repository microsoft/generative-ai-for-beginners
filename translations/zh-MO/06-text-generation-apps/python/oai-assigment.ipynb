## 什麼是文字生成應用程式？

通常當你建立應用程式時，會有某種介面像如下：

- 基於指令。控制台應用程式是典型的指令型應用，你輸入指令並執行任務。例如，`git` 就是指令型的應用程式。
- 使用者介面（UI）。有些應用程式有圖形使用者介面（GUI），你可以點擊按鈕、輸入文字、選擇選項等。

### 控制台和使用者介面應用程式的限制

與指令型應用程式相比，你需要輸入指令：

- <strong>有限制</strong>。你不能隨意輸入任何指令，只能使用應用程式支援的指令。
- <strong>語言特定</strong>。有些應用程式支援多種語言，但預設是為特定語言建立，雖然你可以添加其它語言支援。

### 文字生成應用程式的優點

那麼文字生成應用程式有什麼不同呢？

在文字生成應用程式中，你擁有更大的彈性，不受限於一組指令或特定的輸入語言。相反地，你可以使用自然語言與應用程式互動。另一個好處是，你正在與一個已經從龐大資料庫訓練出來的資料來源互動，而傳統應用程式可能僅限於資料庫中的內容。

### 我可以用文字生成應用程式做什麼？

你可以建立許多不同的應用程式。例如：

- <strong>聊天機器人</strong>。一個能夠回答關於主題，例如你的公司及其產品的問題的聊天機器人，是很合適的。
- <strong>輔助工具</strong>。大型語言模型在摘要文字、從文本中獲得見解、產生如履歷表等文本方面表現優異。
- <strong>程式碼助理</strong>。依據你使用的語言模型，你可以建立一個程式碼助理來協助你撰寫程式碼。例如，你可以使用 GitHub Copilot 以及 ChatGPT 這類產品來輔助編碼。

## 產生文字

產生文字的方法是使用 `responses.create` 函數，範例如下：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上述程式碼中，我們建立回應物件並傳入想使用的模型及提示，然後印出生成的文字。

## 練習 - 你的第一個文字生成應用程式

現在我們已學會如何設定及配置 OpenAI 服務，是時候建立你的第一個文字生成應用程式。請按照以下步驟來建立你的應用程式：


1. 建立一個虛擬環境並安裝 openai：

  > [!NOTE] 如果您在 Codespaces 或 Devcontainer 中運行此筆記本，此步驟並非必須


````python
# Create virtual environment
! python -m venv venv
# Activate virtual environment
! source venv/bin/activate
# Install openai package
! pip install openai
````

> [!NOTE]
> 如果你使用的是 Windows，請輸入 `venv\Scripts\activate` 以取代 `source venv/bin/activate`。

> [!NOTE]
> 前往 https://platform.openai.com/settings/organization/api-keys 找到你的 OpenAI 金鑰，並搜尋 `API keys`。你可以在該處建立新的金鑰並立即複製其值。


1. 建立一個 *app.py* 檔案並輸入以下程式碼：


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````


    你應該會見到類似以下的輸出：

    ```output

    ```
