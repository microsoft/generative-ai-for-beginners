## 什麼是文字生成應用程式？

通常，當您建立應用程式時，它會有某種介面，例如：

- 以指令為基礎。控制台應用程式是典型的應用程式，您輸入指令，它執行任務。例如，`git` 是一個以指令為基礎的應用程式。
- 使用者介面（UI）。有些應用程式有圖形使用者介面（GUI），您可以點擊按鈕、輸入文字、選擇選項等等。

### 控制台和 UI 應用程式的限制

與您輸入指令的指令式應用比較：

- <strong>有其限制</strong>。您不能只是輸入任何指令，只有該應用程式支持的指令才行。
- <strong>語言特定</strong>。有些應用程式支持多種語言，但預設是針對特定語言建立，即使您可以添加更多語言支持。

### 文字生成應用程式的優點

那麼，文字生成應用程式有什麼不同呢？

在文字生成應用程式中，您擁有更多彈性，不受限於一組指令或特定輸入語言。相反，您可以使用自然語言與應用程式互動。另一個好處是，因為您已經在與經過大量語料庫訓練的資料源互動，而傳統應用程式可能僅限於資料庫中的內容。

### 我能用文字生成應用程式做什麼？

有許多東西您可以建立。例如：

- <strong>聊天機器人</strong>。用於回答關於主題的問題，例如您的公司及其產品，會是很好的應用。
- <strong>助理</strong>。大型語言模型（LLM）擅長文字摘要、從文字中獲得洞見、產生文字如履歷等。
- <strong>程式碼助理</strong>。根據您使用的語言模型，您可以建立程式碼助理協助您寫程式碼。例如，您可以使用 GitHub Copilot 及 ChatGPT 這類產品來幫助編寫程式碼。

## 產生文字

產生文字的方式是使用 `responses.create` 方法。以下是一個範例：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上面的程式碼中，我們建立一個回應對象，並傳入想使用的模型和提示。然後印出生成的文字。

### 聊天回應

現在我們已學會如何設置和配置 OpenAI 服務，該是建立您第一個文字生成應用程式的時候了。請依照以下步驟建立您的應用程式：


1. 建立虛擬環境並安裝 openai：

  > [!NOTE] 如果您在 Codespaces 或 Devcontainer 中執行此筆記本，則此步驟不是必需的

1. 建立一個 *app.py* 檔案並寫入以下代碼：


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

# add your completion code
prompt = "Complete the following: Once upon a time there was a"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, store=False)

# print response
print(response.output_text)
````
