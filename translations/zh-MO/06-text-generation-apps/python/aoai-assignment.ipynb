## 什麼是文本生成應用程式？

通常當你建立一個應用程式，它會有某種類型的介面，例如：

- 基於指令的。命令列應用程式是典型的應用，你輸入指令，它就執行任務。例如，`git` 是基於指令的應用程式。
- 使用者介面（UI）。有些應用有圖形使用者介面（GUI），你點擊按鈕、輸入文字、選擇選項等。

### 命令列和 UI 應用程式的限制

與基於指令的應用相比，你必須輸入指令：

- <strong>它有限制</strong>。你不能輸入任意指令，僅限於應用程式支援的指令。
- <strong>語言特定</strong>。有些應用程式支援多種語言，但預設情況下，應用是為特定語言打造，即使你可以新增更多語言支援。

### 文本生成應用程式的好處

那麼，文本生成應用程式有什麼不同呢？

在文本生成應用程式中，你擁有更大的彈性，不受限於固定的指令集或特定的輸入語言。相反，你可以使用自然語言與應用程式互動。另一個優點是你已經在和一個經過大量語料庫訓練的資料源進行互動，而傳統應用程式可能只限於資料庫中的資料。

### 我可以用文本生成應用程式建構什麼？

你可以建立許多不同的應用。例如：

- <strong>聊天機器人</strong>。像是回答有關你公司及其產品的問題的聊天機器人會是一個不錯的選擇。
- <strong>助手</strong>。大型語言模型（LLM）非常擅長摘要文字、從文本取得洞見、產生如履歷等文本內容。
- <strong>程式碼助理</strong>。根據你使用的語言模型，可以建立協助撰寫程式碼的助理。例如，你可以使用 GitHub Copilot 或 ChatGPT 幫助你寫程式碼。

### 設定 Azure 配置

如果你使用的是 Azure OpenAI，以下是配置設定方式。Responses API 從 Azure OpenAI **v1 端點** 提供服務，因此我們將 `OpenAI` 客戶端指向 `<your-endpoint>/openai/v1/`：

```python

  api_key=os.environ['AZURE_OPENAI_API_KEY'],
```

- `api_key`，即你在 Azure 入口網站得到的 API 金鑰。

> [!NOTE]

## 產生文本

產生文本的方法是使用 `responses.create`，以下是範例：

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

在上述程式中，我們建立了一個回應物件，傳入要使用的模型與提示詞。然後我們印出產生的文本。
