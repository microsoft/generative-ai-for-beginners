## Cos'è un'app per la generazione di testo?

Normalmente quando costruisci un'app ha qualche tipo di interfaccia come la seguente:

- Basata su comandi. Le app da console sono app tipiche in cui digiti un comando e questo esegue un compito. Ad esempio, `git` è un'app basata su comandi.
- Interfaccia utente (UI). Alcune app hanno interfacce grafiche (GUI) dove clicchi pulsanti, inserisci testo, selezioni opzioni e altro.

### Le app da console e UI sono limitate

Confrontalo con un'app basata su comandi dove digiti un comando:

- **È limitata**. Non puoi semplicemente digitare qualsiasi comando, solo quelli che l'app supporta.
- **Specificità linguistica**. Alcune app supportano molte lingue, ma di default l'app è costruita per una lingua specifica, anche se puoi aggiungere supporto per altre lingue.

### Vantaggi delle app per la generazione di testo

Allora, in cosa è diversa un'app per la generazione di testo?

In un'app per la generazione di testo, hai più flessibilità, non sei limitato a un insieme di comandi o a una lingua specifica di input. Invece, puoi usare il linguaggio naturale per interagire con l'app. Un altro vantaggio è che poiché stai già interagendo con una fonte dati che è stata addestrata su un vasto corpus di informazioni, mentre un'app tradizionale potrebbe essere limitata a ciò che è in un database.

### Cosa posso costruire con un'app per la generazione di testo?

Ci sono molte cose che puoi costruire. Per esempio:

- **Un chatbot**. Un chatbot che risponde a domande su argomenti come la tua azienda e i suoi prodotti potrebbe essere una buona soluzione.
- **Assistente**. I LLM sono ottimi per cose come riassumere testi, ottenere insight dal testo, produrre testi come curriculum e altro.
- **Assistente di codice**. A seconda del modello di linguaggio che usi, puoi costruire un assistente di codice che ti aiuta a scrivere codice. Per esempio, puoi usare un prodotto come GitHub Copilot così come ChatGPT per aiutarti a scrivere codice.

## Genera testo

Il modo per generare testo è usare il metodo `responses.create`. Ecco un esempio:

```python

```

Nel codice sopra, creiamo un oggetto risposta e passiamo il modello che vogliamo usare e il prompt. Poi stampiamo il testo generato.

## Esercizio - la tua prima app di generazione di testo

Ottimo, vediamo come possiamo migliorare le cose. Per migliorare le cose, vogliamo assicurarci che il codice sia flessibile, così che gli ingredienti e il numero di ricette possano essere migliorati e modificati. 


1. Cambiamo il codice nel modo seguente:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Qui suggeriamo di aggiungere un prompt aggiuntivo, ma per farlo funzionare, dobbiamo aggiungere il risultato del primo prompt come contesto al secondo prompt.

    Trova la parte del codice che stampa il risultato del primo prompt e aggiungi il seguente codice sotto:
    
    ```python
    old_prompt_result = response.output_text
    prompt = "Produce a shopping list for the generated recipes and please don't include ingredients that I already have."
    
    new_prompt = f"{old_prompt_result} {prompt}"
    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    
    # stampa risposta
    print("Shopping list:")
    print(response.output_text)
    ```

    - Facciamo una nuova richiesta, ma considerando anche il numero di token richiesti nel primo prompt, quindi questa volta diciamo che `max_output_tokens` è 1200.

- **Una parola sulla lunghezza dei token**. Dovremmo considerare quanti token sono necessari per generare il testo che vogliamo. I token hanno un costo, quindi dove possibile dovremmo cercare di essere economici con il numero di token che usiamo. Ad esempio, possiamo formulare il prompt in modo da usare meno token?

   Per cambiare i token usati, puoi usare il parametro `max_output_tokens`. Ad esempio, se vuoi usare 100 token, faresti così:


- **Sperimentare con la temperatura**. La temperatura è qualcosa che non abbiamo ancora menzionato ma è un contesto importante per come il nostro programma si comporta. Più alto è il valore della temperatura, più casuale sarà l'output. Al contrario, più basso è il valore della temperatura, più prevedibile sarà l'output. Considera se vuoi variazione nel tuo output o no.

   Per modificare la temperatura, puoi usare il parametro `temperature`. Ad esempio, se vuoi usare una temperatura di 0,5, faresti così:

    ```python
    response = client.responses.create(model=deployment, input=messages, store=False)
    ```
