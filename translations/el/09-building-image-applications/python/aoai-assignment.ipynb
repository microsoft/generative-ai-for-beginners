## Τι είναι το DALL-E και το Midjourney; 

Το [DALL-E](https://openai.com/dall-e-2?WT.mc_id=academic-105485-koreyst) και το [Midjourney](https://www.midjourney.com/?WT.mc_id=academic-105485-koreyst) είναι δύο από τα πιο δημοφιλή μοντέλα δημιουργίας εικόνων, που σας επιτρέπουν να χρησιμοποιήσετε εντολές για να δημιουργήσετε εικόνες.

### DALL-E

Ας ξεκινήσουμε με το DALL-E, το οποίο είναι ένα μοντέλο Γενετικής Τεχνητής Νοημοσύνης που δημιουργεί εικόνες από περιγραφές κειμένου. 

> [Το DALL-E είναι ένας συνδυασμός δύο μοντέλων, του CLIP και της διαχυμένης προσοχής](https://towardsdatascience.com/openais-dall-e-and-clip-101-a-brief-introduction-3a4367280d4e?WT.mc_id=academic-105485-koreyst).  

- **Το CLIP** είναι ένα μοντέλο που δημιουργεί ενσωματώσεις (embeddings), που είναι αριθμητικές αναπαραστάσεις δεδομένων, από εικόνες και κείμενο.  

- **Η διαχυμένη προσοχή (diffused attention)** είναι ένα μοντέλο που δημιουργεί εικόνες από ενσωματώσεις. Το DALL-E εκπαιδεύτηκε σε ένα σύνολο δεδομένων με εικόνες και κείμενο και μπορεί να χρησιμοποιηθεί για να δημιουργήσει εικόνες από περιγραφές κειμένου. Για παράδειγμα, το DALL-E μπορεί να χρησιμοποιηθεί για να δημιουργήσει εικόνες από μια γάτα με καπέλο ή ένα σκύλο με μοϊκάνα. 

### Midjourney
 
Το Midjourney λειτουργεί παρόμοια με το DALL-E, δημιουργεί εικόνες από εντολές κειμένου. Το Midjourney, μπορεί επίσης να χρησιμοποιηθεί για τη δημιουργία εικόνων με εντολές όπως "μια γάτα με καπέλο" ή "ένας σκύλος με μοϊκάνα".

## Πώς λειτουργούν το DALL-E και το Midjourney 

Πρώτα, το [DALL-E](https://arxiv.org/pdf/2102.12092.pdf?WT.mc_id=academic-105485-koreyst). Το DALL-E είναι ένα μοντέλο Γενετικής Τεχνητής Νοημοσύνης βασισμένο στην αρχιτεκτονική μετασχηματιστών με *αυτοπαραγοντικό μετασχηματιστή* (autoregressive transformer).  

Ένας *αυτοπαραγοντικός μετασχηματιστής* καθορίζει πώς ένα μοντέλο δημιουργεί εικόνες από περιγραφές κειμένου, δημιουργεί ένα pixel τη φορά και μετά χρησιμοποιεί τα δημιουργημένα pixels για να δημιουργήσει το επόμενο pixel. Διαπερνώντας πολλαπλές στρώσεις σε ένα νευρωνικό δίκτυο, μέχρι να ολοκληρωθεί η εικόνα.  

Με αυτή τη διαδικασία, το DALL-E ελέγχει χαρακτηριστικά, αντικείμενα, ιδιότητες και άλλα στην εικόνα που δημιουργεί. Ωστόσο, το DALL-E 2 και 3 έχουν περισσότερο έλεγχο πάνω στην παραγόμενη εικόνα, 


## Δημιουργία της πρώτης σας εφαρμογής δημιουργίας εικόνων

Τι χρειάζεται για να δημιουργήσετε μια εφαρμογή δημιουργίας εικόνων; Χρειάζεστε τις παρακάτω βιβλιοθήκες:

- **python-dotenv**, σας συνιστάται έντονα να χρησιμοποιήσετε αυτή τη βιβλιοθήκη για να κρατάτε τα μυστικά σας σε ένα αρχείο *.env* μακριά από τον κώδικα.
- **openai**, αυτή η βιβλιοθήκη είναι που θα χρησιμοποιήσετε για να αλληλεπιδράσετε με το API του OpenAI.
- **pillow**, για να δουλέψετε με εικόνες σε Python.
- **requests**, για να σας βοηθήσει να κάνετε αιτήματα HTTP.

Αν δεν το έχετε κάνει ήδη, ακολουθήστε τις οδηγίες στη σελίδα [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) για να δημιουργήσετε έναν πόρο Azure OpenAI και ένα μοντέλο. Επιλέξτε **gpt-image-1** ως μοντέλο (το τρέχον μοντέλο δημιουργίας εικόνων Azure OpenAI· το DALL-E 3 είναι παλαιότερης γενιάς και δεν είναι πλέον διαθέσιμο για νέες αναπτύξεις).

1. Προσθέστε τον παρακάτω κώδικα σε αρχείο που ονομάζεται *app.py*:

    ```python
    import openai
    import os
    import requests
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI

    # εισαγωγή dotenv
    dotenv.load_dotenv()

    # διαμόρφωση πελάτη υπηρεσίας Azure OpenAI
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2024-10-21"
      )

    try:
        # Δημιουργήστε μια εικόνα χρησιμοποιώντας το API δημιουργίας εικόνων
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Εισαγάγετε το κείμενο του προτροπέα σας εδώ
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']
        )
        # Ορίστε τον κατάλογο για την αποθηκευμένη εικόνα
        image_dir = os.path.join(os.curdir, 'images')
    
        # Εάν ο κατάλογος δεν υπάρχει, δημιουργήστε τον
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)
    
        # Αρχικοποιήστε τη διαδρομή της εικόνας (σημειώστε ότι ο τύπος αρχείου πρέπει να είναι png)
        image_path = os.path.join(image_dir, 'generated-image.png')
    
        # Ανακτήστε την παραγόμενη εικόνα
        image_url = generation_response.data[0].url  # εξαγάγετε το URL της εικόνας από την απόκριση
        generated_image = requests.get(image_url).content  # κατεβάστε την εικόνα
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)
    
        # Εμφανίστε την εικόνα στην προεπιλεγμένη προβολή εικόνων
        image = Image.open(image_path)
        image.show()
    
    # σύλληψη εξαιρέσεων
    except openai.BadRequestError as err:
        print(err)

    ```

### Περισσότερες λεπτομέρειες για τη δημιουργία της εικόνας

Ας δούμε τον κώδικα που δημιουργεί την εικόνα με περισσότερη λεπτομέρεια:

```python
generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Εισάγετε το κείμενο προτροπής σας εδώ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
```

- **prompt**, είναι το κείμενο-εντολή που χρησιμοποιείται για τη δημιουργία της εικόνας. Σε αυτή την περίπτωση, χρησιμοποιούμε το prompt "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- **size**, είναι το μέγεθος της εικόνας που δημιουργείται. Σε αυτή την περίπτωση, δημιουργούμε μια εικόνα 1024x1024 pixels.
- **n**, είναι ο αριθμός των εικόνων που δημιουργούνται. Σε αυτή την περίπτωση, δημιουργούμε μία εικόνα.

Υπάρχουν περισσότερα πράγματα που μπορείτε να κάνετε με τις εικόνες και θα τα καλύψουμε στην επόμενη ενότητα.
