# Δημιουργία Εφαρμογών Παραγωγής Εικόνων (Azure OpenAI)

Υπάρχει περισσότερα στα LLMs από την παραγωγή κειμένου. Μπορείτε επίσης να δημιουργήσετε εικόνες από περιγραφές κειμένου. Οι εικόνες ως μέσο είναι χρήσιμες σε τομείς όπως η MedTech, η αρχιτεκτονική, ο τουρισμός, η ανάπτυξη παιχνιδιών, το μάρκετινγκ και άλλα. Σε αυτό το μάθημα δουλεύουμε με τα σημερινά μοντέλα **GPT Image** της Microsoft Foundry.

## Στόχοι μάθησης

Μέχρι το τέλος αυτού του μαθήματος θα μπορείτε να:

- Εξηγήσετε τι είναι η παραγωγή εικόνων και πού είναι χρήσιμη.
- Κατανοήσετε την οικογένεια μοντέλων `gpt-image` και πώς διαφέρει από τα παλαιότερα μοντέλα DALL·E.
- Δημιουργήσετε μια εφαρμογή παραγωγής εικόνων και να επεξεργαστείτε εικόνες.

## Τι είναι η παραγωγή εικόνων;

Τα μοντέλα παραγωγής εικόνων δημιουργούν εικόνες από ένα κειμενικό προτροπή. Τα σύγχρονα μοντέλα όπως το `gpt-image` μαθαίνουν τη σχέση μεταξύ κειμένου και εικόνων κατά την εκπαίδευση και, στη συνέχεια, μετατρέπουν επαναληπτικά τον τυχαίο θόρυβο σε μια εικόνα που ταιριάζει στο προτροπή σας.

Δύο γνωστές οικογένειες μοντέλων εικόνων είναι:

- **`gpt-image` (OpenAI)** — η τρέχουσα γενιά που χρησιμοποιείται σε αυτό το μάθημα. Υποστηρίζει παραγωγή εικόνας από κείμενο και επεξεργασία εικόνας (inpainting με μάσκα).
- **Midjourney** — ένα δημοφιλές τρίτο μοντέλο με δική του υπηρεσία και ροή εργασίας βασισμένη στο Discord.

> Τα παλαιότερα μοντέλα εικόνας της OpenAI — **DALL·E 2** και **DALL·E 3** — είναι παρωχημένα. Το DALL·E 3 δεν είναι πλέον διαθέσιμο για νέες αναπτύξεις, και λειτουργίες όπως το `create_variation` υπήρχαν μόνο στο DALL·E 2. Χρησιμοποιήστε τα μοντέλα `gpt-image` για νέες εφαρμογές.

Στη Microsoft Foundry, το **`gpt-image-2`** είναι το πιο πρόσφατο και πιο ικανό μοντέλο εικόνας και είναι η προτεινόμενη προεπιλογή. Τα `gpt-image-1.5` και `gpt-image-1-mini` είναι επίσης γενικά διαθέσιμα.

> **Σημαντικό:** Τα μοντέλα `gpt-image` επιστρέφουν την παραγόμενη εικόνα ως **base64** (`b64_json`), όχι ως URL. Ο κώδικάς σας αποκωδικοποιεί τη συμβολοσειρά base64 σε bytes και την αποθηκεύει — δεν υπάρχει URL εικόνας για λήψη.


## Δημιουργία της πρώτης εφαρμογής παραγωγής εικόνας

Τι χρειάζεται λοιπόν για να δημιουργήσετε μια εφαρμογή παραγωγής εικόνας; Χρειάζεστε τις ακόλουθες βιβλιοθήκες:

- **python-dotenv**, συνιστάται ιδιαίτερα να χρησιμοποιήσετε αυτή τη βιβλιοθήκη για να κρατάτε τα μυστικά σας σε ένα αρχείο *.env* μακριά από τον κώδικα.
- **openai**, αυτή η βιβλιοθήκη είναι αυτή που θα χρησιμοποιήσετε για να αλληλεπιδράσετε με το API της OpenAI.
- **pillow**, για να δουλεύετε με εικόνες σε Python.

Εάν δεν το έχετε ήδη κάνει, ακολουθήστε τις οδηγίες στη σελίδα [Microsoft Learn](https://learn.microsoft.com/azure/ai-foundry/openai/how-to/create-resource?pivots=web-portal&WT.mc_id=academic-105485-koreyst) για να δημιουργήσετε έναν πόρο Microsoft Foundry και μοντέλο. Επιλέξτε το **gpt-image-2** ως το μοντέλο (το πιο πρόσφατο μοντέλο εικόνας Azure OpenAI· το DALL·E είναι παλαιότερο).

1. Δημιουργήστε ένα αρχείο *.env* με τα ακόλουθα περιεχόμενα:

    ```text
    AZURE_OPENAI_ENDPOINT=<your endpoint>
    AZURE_OPENAI_API_KEY=<your key>
    AZURE_OPENAI_DEPLOYMENT="gpt-image-2"
    ```

    Εντοπίστε αυτές τις πληροφορίες στην πύλη Microsoft Foundry για τον πόρο σας στην ενότητα "Deployments".



1. Συλλέξτε τις παραπάνω βιβλιοθήκες σε ένα αρχείο με όνομα *requirements.txt* ως εξής:

    ```text
    python-dotenv
    openai
    pillow
    ```


1. Στη συνέχεια, δημιουργήστε ένα εικονικό περιβάλλον και εγκαταστήστε τις βιβλιοθήκες:


`````python
# create virtual env
! python3 -m venv venv
# activate environment
! source venv/bin/activate
# install libraries
# pip install -r requirements.txt, if using a requirements.txt file 
! pip install python-dotenv openai pillow
`````

> [!NOTE]
> Για τα Windows, χρησιμοποιήστε τις ακόλουθες εντολές για να δημιουργήσετε και να ενεργοποιήσετε το εικονικό σας περιβάλλον:

    ```bash
    python3 -m venv venv
    venv\Scripts\activate.bat
    ````

1. Προσθέστε τον παρακάτω κώδικα σε αρχείο με όνομα *app.py*:

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError

    # εισαγωγή dotenv
    dotenv.load_dotenv()

    # διαμόρφωση πελάτη Azure OpenAI (Microsoft Foundry).
    # Τα μοντέλα εικόνας χρειάζονται πρόσφατη έκδοση API — ελέγξτε τα έγγραφα Foundry για αυτή που απαιτεί το μοντέλο σας.
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )

    try:
        # Δημιουργήστε μια εικόνα χρησιμοποιώντας το API δημιουργίας εικόνων
        generation_response = client.images.generate(
            prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Εισάγετε εδώ το κείμενο της υπόδειξής σας
            size='1024x1024',
            n=1,
            model=os.environ['AZURE_OPENAI_DEPLOYMENT']   # π.χ. "gpt-image-2"
        )
        # Ορίστε τον κατάλογο για την αποθηκευμένη εικόνα
        image_dir = os.path.join(os.curdir, 'images')

        # Εάν ο κατάλογος δεν υπάρχει, δημιουργήστε τον
        if not os.path.isdir(image_dir):
            os.mkdir(image_dir)

        # Αρχικοποιήστε τη διαδρομή εικόνας (σημειώστε ότι ο τύπος αρχείου πρέπει να είναι png)
        image_path = os.path.join(image_dir, 'generated-image.png')

        # τα μοντέλα gpt-image επιστρέφουν την εικόνα ως base64 (b64_json), όχι ως URL
        image_b64 = generation_response.data[0].b64_json
        generated_image = base64.b64decode(image_b64)
        with open(image_path, "wb") as image_file:
            image_file.write(generated_image)

        # Εμφάνιση της εικόνας στον προεπιλεγμένο προβολέα εικόνων
        image = Image.open(image_path)
        image.show()

    # αναφορές εξαιρέσεων
    except BadRequestError as err:
        print(err)

    ```

Ας εξηγήσουμε αυτόν τον κώδικα:

- Πρώτα, εισάγουμε τις βιβλιοθήκες που χρειαζόμαστε, συμπεριλαμβανομένης της βιβλιοθήκης OpenAI, της βιβλιοθήκης dotenv, του module base64 και της βιβλιοθήκης Pillow.

    ```python
    import os
    import base64
    from PIL import Image
    import dotenv
    from openai import AzureOpenAI, BadRequestError
    ```

- Στη συνέχεια, φορτώνουμε τις μεταβλητές περιβάλλοντος από το αρχείο *.env*.

    ```python
    # εισαγωγή dotenv
    dotenv.load_dotenv()
    ```

- Μετά από αυτό, διαμορφώνουμε τον πελάτη Azure OpenAI (Microsoft Foundry).

    ```python
    client = AzureOpenAI(
      azure_endpoint = os.environ["AZURE_OPENAI_ENDPOINT"],
      api_key=os.environ['AZURE_OPENAI_API_KEY'],
      api_version = "2025-04-01-preview"
      )
    ```

- Έπειτα, παράγουμε την εικόνα:

    ```python
    generation_response = client.images.generate(
        prompt='Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils',    # Εισάγετε το κείμενο του προτροπής σας εδώ
        size='1024x1024',
        n=1,
        model=os.environ['AZURE_OPENAI_DEPLOYMENT']
    )
    ```

    Τα μοντέλα `gpt-image` επιστρέφουν την εικόνα ως συμβολοσειρά **base64** στο `data[0].b64_json`. Την αποκωδικοποιούμε σε bytes και την εγγράφουμε σε αρχείο — δεν υπάρχει URL για λήψη.

    ```python
    image_b64 = generation_response.data[0].b64_json
    generated_image = base64.b64decode(image_b64)
    with open(image_path, "wb") as image_file:
        image_file.write(generated_image)
    ```

- Τέλος, ανοίγουμε την εικόνα και χρησιμοποιούμε τον τυπικό προβολέα εικόνων για να την εμφανίσουμε:

    ```python
    image = Image.open(image_path)
    image.show()
    ```

### Περισσότερες λεπτομέρειες για τη δημιουργία της εικόνας

Ας δούμε τις παραμέτρους του `images.generate`:

- Η **προτροπή** (prompt), είναι το κείμενο που χρησιμοποιείται για να δημιουργηθεί η εικόνα. Εδώ είναι "Bunny on horse, holding a lollipop, on a foggy meadow where it grows daffodils".
- Το **μέγεθος** (size), είναι το μέγεθος της παραγόμενης εικόνας (για παράδειγμα `1024x1024`, `1536x1024`, `1024x1536`, ή `"auto"`).
- Ο **αριθμός** (n), είναι ο αριθμός των παραγόμενων εικόνων. Εδώ παράγουμε μία.
- Το **μοντέλο** (model), είναι το όνομα της ανάπτυξης του μοντέλου εικόνας σας (για παράδειγμα `gpt-image-2`).

> Τα μοντέλα εικόνας δεν λαμβάνουν παράμετρο `temperature` — αυτή είναι έλεγχος για τη δημιουργία κειμένου. Για να έχετε ποικιλία, καλέστε ξανά το API· για να μειώσετε την ποικιλία, κάντε την προτροπή σας πιο συγκεκριμένη.

## Επιπλέον δυνατότητες δημιουργίας εικόνας

Έχετε δει πώς να δημιουργήσετε μια εικόνα με λίγες γραμμές Python. Τα μοντέλα `gpt-image` μπορούν επίσης να **επεξεργαστούν** μια υπάρχουσα εικόνα. Παρέχοντας μια εικόνα, μια προαιρετική **μάσκα** (που σημειώνει την περιοχή που θα αλλάξει) και μια προτροπή, μπορείτε να τροποποιήσετε μέρος μιας εικόνας — για παράδειγμα, να προσθέσετε ένα καπέλο στον λαγό μας.

```python
response = client.images.edit(
  model=os.environ['AZURE_OPENAI_DEPLOYMENT'],
  image=open("base_image.png", "rb"),
  mask=open("mask.png", "rb"),
  prompt="An image of a rabbit with a hat on its head.",
  n=1,
  size="1024x1024"
)
# οι επεξεργασίες επιστρέφονται επίσης ως base64
edited_image = base64.b64decode(response.data[0].b64_json)
```

Η βασική εικόνα περιέχει μόνο το λαγό· η τελική εικόνα έχει το καπέλο στον λαγό.
