## Τι είναι η εφαρμογή δημιουργίας κειμένου;

Κανονικά όταν δημιουργείτε μια εφαρμογή, έχει κάποιο είδος διεπαφής όπως η παρακάτω:

- Βασιζόμενη σε εντολές. Οι κονσόλες εφαρμογών είναι τυπικές εφαρμογές όπου πληκτρολογείτε μια εντολή και εκτελεί μια εργασία. Για παράδειγμα, το `git` είναι μια εφαρμογή βασισμένη σε εντολές.
- Διεπαφή χρήστη (UI). Ορισμένες εφαρμογές έχουν γραφικές διεπαφές χρήστη (GUIs) όπου κάνετε κλικ σε κουμπιά, εισάγετε κείμενο, επιλέγετε επιλογές και άλλα.

### Οι εφαρμογές κονσόλας και UI έχουν περιορισμούς

Συγκρίνετέ το με μια εφαρμογή βασισμένη σε εντολές όπου πληκτρολογείτε μια εντολή: 

- **Είναι περιορισμένη**. Δεν μπορείτε απλά να πληκτρολογήσετε όποια εντολή θέλετε, μόνο αυτές που υποστηρίζει η εφαρμογή.
- **Ειδική γλώσσα**. Κάποιες εφαρμογές υποστηρίζουν πολλές γλώσσες, αλλά από προεπιλογή η εφαρμογή δημιουργείται για μια συγκεκριμένη γλώσσα, ακόμα και αν μπορείτε να προσθέσετε υποστήριξη για περισσότερες γλώσσες. 

### Οφέλη των εφαρμογών δημιουργίας κειμένου

Πώς διαφέρει λοιπόν μια εφαρμογή δημιουργίας κειμένου;

Αντ' αυτού, μπορείτε να χρησιμοποιήσετε φυσική γλώσσα για να αλληλεπιδράσετε με την εφαρμογή. ## Δημιουργία κειμένου

Ο τρόπος για να δημιουργήσετε κείμενο είναι να χρησιμοποιήσετε τη μέθοδο `responses.create`. Να ένα παράδειγμα:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Στον παραπάνω κώδικα, δημιουργούμε ένα αντικείμενο απόκρισης και περνάμε το μοντέλο που θέλουμε να χρησιμοποιήσουμε και την υπόδειξη. Έπειτα εκτυπώνουμε το παραγόμενο κείμενο.

### Απαντήσεις συνομιλίας

Μέχρι στιγμής, είδατε πώς μπορούμε να δημιουργήσουμε κείμενο από μια μοναδική υπόδειξη. Το Responses API είναι επίσης κατάλληλο για chatbots – παρέχετε μια λίστα μηνυμάτων ως `input` για να χτίσετε μια συνομιλία. Να ένα παράδειγμα για τη χρήση του:

```python
client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
  )

deployment = "gpt-5-mini"

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Περισσότερα γι' αυτή τη λειτουργικότητα σε επόμενο κεφάλαιο.

## Άσκηση - Δημιουργία γεννήτριας συνταγών

Τώρα που έχουμε παίξει ένα σενάριο, ας γράψουμε κώδικα που να ταιριάζει με το επιδεικνυόμενο σενάριο. Για να το κάνετε αυτό, ακολουθήστε τα παρακάτω βήματα:

1. Χρησιμοποιήστε το υπάρχον αρχείο *app.py* ως σημείο εκκίνησης
1. Εντοπίστε τη μεταβλητή `prompt` και αλλάξτε τον κώδικά της ως εξής:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> ΣΗΜΕΙΩΣΗ, το LLM σας είναι μη ντετερμινιστικό, οπότε μπορεί να παίρνετε διαφορετικά αποτελέσματα κάθε φορά που εκτελείτε το πρόγραμμα.

Τέλεια, ας δούμε πώς μπορούμε να βελτιώσουμε τα πράγματα. Για να βελτιώσουμε τα πράγματα, θέλουμε να διασφαλίσουμε ότι ο κώδικας είναι ευέλικτος, ώστε τα υλικά και ο αριθμός των συνταγών να μπορούν να βελτιωθούν και να αλλάξουν. 


1. Ας αλλάξουμε τον κώδικα με τον ακόλουθο τρόπο:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-5-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Έναρξη του κώδικα για δοκιμή, μπορεί να μοιάζει έτσι:

    old_prompt_result = response.output_text
