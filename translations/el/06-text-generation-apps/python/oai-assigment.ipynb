## Τι είναι μια εφαρμογή δημιουργίας κειμένου;

Κανονικά όταν φτιάχνετε μια εφαρμογή έχει κάποια μορφή διεπαφής όπως η εξής:

## Άσκηση - δημιουργία γεννήτριας συνταγών

Τώρα που έχουμε παίξει ένα σενάριο, ας γράψουμε κώδικα που να ταιριάζει με το επιδειχθέν σενάριο. Για να το κάνετε αυτό, ακολουθήστε τα βήματα:

1. Χρησιμοποιήστε το υπάρχον αρχείο *app.py* ως σημείο εκκίνησης
1. Εντοπίστε τη μεταβλητή `prompt` και αλλάξτε τον κώδικά της στον εξής: 


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

> ΣΗΜΕΙΩΣΗ, το LLM σας είναι μη ντετερμινιστικό, οπότε μπορεί να παίρνετε διαφορετικά αποτελέσματα κάθε φορά που τρέχετε το πρόγραμμα.

Τέλεια, ας δούμε πώς μπορούμε να βελτιώσουμε τα πράγματα. Για να βελτιώσουμε τα πράγματα, θέλουμε να διασφαλίσουμε ότι ο κώδικας είναι ευέλικτος, έτσι ώστε τα υλικά και ο αριθμός των συνταγών να μπορούν να βελτιώνονται και να αλλάζουν. 


1. Ας αλλάξουμε τον κώδικα με τον ακόλουθο τρόπο:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key = os.environ['OPENAI_API_KEY']
)

deployment = "gpt-4o-mini"

no_recipes = input("No of recipes (for example, 5: ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots: ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Εδώ προτείνουμε να προσθέσουμε ένα επιπλέον prompt, αλλά για να λειτουργήσει αυτό, πρέπει να προσθέσουμε το αποτέλεσμα του πρώτου prompt ως πλαίσιο στο δεύτερο prompt.

    Βρείτε το μέρος στον κώδικα που εκτυπώνει το αποτέλεσμα από το πρώτο prompt και προσθέστε τον παρακάτω κώδικα από κάτω:
    
    ```python
    old_prompt_result = response.output_text

    messages = [{"role": "user", "content": new_prompt}]
    response = client.responses.create(model=deployment, input=messages, max_output_tokens=1200, store=False)
    ```
