# Δημιουργία εφαρμογών παραγωγής κειμένου

Μέχρι τώρα σε αυτό το πρόγραμμα σπουδών έχεις δει βασικές έννοιες όπως τα prompts και ακόμα και μια ολόκληρη επιστήμη που ονομάζεται "prompt engineering". Πολλά εργαλεία με τα οποία μπορείς να αλληλεπιδράσεις, όπως το ChatGPT, το Office 365, το Microsoft Power Platform και άλλα, σε βοηθούν να χρησιμοποιείς prompts για να πετύχεις κάτι.

Για να προσθέσεις μια τέτοια εμπειρία σε μια εφαρμογή, χρειάζεται να κατανοήσεις έννοιες όπως prompts, completions και να επιλέξεις μια βιβλιοθήκη για να δουλέψεις. Αυτό ακριβώς θα μάθεις σε αυτό το κεφάλαιο.

- Κατανοήσεις πώς να χρησιμοποιείς έννοιες όπως prompt, temperature και tokens για να φτιάξεις μια εφαρμογή παραγωγής κειμένου.

## Στόχοι μάθησης

Στο τέλος αυτού του μαθήματος, θα μπορείς να:

## Τι είναι μια εφαρμογή παραγωγής κειμένου;

Συνήθως όταν φτιάχνεις μια εφαρμογή, έχει κάποιο είδος διεπαφής όπως τα παρακάτω:

- Βασισμένη σε εντολές. Οι εφαρμογές κονσόλας είναι τυπικές εφαρμογές όπου πληκτρολογείς μια εντολή και εκτελείται μια εργασία. Για παράδειγμα, το `git` είναι μια εφαρμογή βασισμένη σε εντολές.
- Γραφικό περιβάλλον χρήστη (UI). Κάποιες εφαρμογές έχουν γραφικό περιβάλλον όπου κάνεις κλικ σε κουμπιά, εισάγεις κείμενο, επιλέγεις επιλογές κ.ά.

### Οι εφαρμογές κονσόλας και UI έχουν περιορισμούς

Σύγκρινέ το με μια εφαρμογή βασισμένη σε εντολές όπου πληκτρολογείς μια εντολή:

- **Έχει περιορισμούς**. Δεν μπορείς να πληκτρολογήσεις οποιαδήποτε εντολή, μόνο αυτές που υποστηρίζει η εφαρμογή.
- **Εξαρτάται από τη γλώσσα**. Κάποιες εφαρμογές υποστηρίζουν πολλές γλώσσες, αλλά από προεπιλογή η εφαρμογή είναι φτιαγμένη για μια συγκεκριμένη γλώσσα, ακόμα κι αν μπορείς να προσθέσεις υποστήριξη για άλλες.

### Πλεονεκτήματα των εφαρμογών παραγωγής κειμένου

Πώς διαφέρει λοιπόν μια εφαρμογή παραγωγής κειμένου;

Σε μια τέτοια εφαρμογή έχεις περισσότερη ευελιξία, δεν περιορίζεσαι σε ένα σύνολο εντολών ή σε μια συγκεκριμένη γλώσσα εισόδου. Αντίθετα, μπορείς να χρησιμοποιήσεις φυσική γλώσσα για να αλληλεπιδράσεις με την εφαρμογή. Ένα ακόμα πλεονέκτημα είναι ότι αλληλεπιδράς ήδη με μια πηγή δεδομένων που έχει εκπαιδευτεί σε τεράστιο όγκο πληροφορίας, ενώ μια παραδοσιακή εφαρμογή μπορεί να περιορίζεται σε ό,τι υπάρχει σε μια βάση δεδομένων.

### Τι μπορώ να φτιάξω με μια εφαρμογή παραγωγής κειμένου;

Υπάρχουν πολλά που μπορείς να δημιουργήσεις. Για παράδειγμα:

- **Ένα chatbot**. Ένα chatbot που απαντά σε ερωτήσεις για θέματα όπως η εταιρεία σου και τα προϊόντα της θα ήταν μια καλή επιλογή.
- **Βοηθός**. Τα LLMs είναι εξαιρετικά σε πράγματα όπως η σύνοψη κειμένου, η εξαγωγή συμπερασμάτων από κείμενο, η παραγωγή κειμένου όπως βιογραφικά και άλλα.
- **Βοηθός κώδικα**. Ανάλογα με το γλωσσικό μοντέλο που χρησιμοποιείς, μπορείς να φτιάξεις έναν βοηθό που σε βοηθά να γράψεις κώδικα. Για παράδειγμα, μπορείς να χρησιμοποιήσεις προϊόντα όπως το GitHub Copilot ή το ChatGPT για να σε βοηθήσουν στο γράψιμο κώδικα.

## Άσκηση - δημιουργία γεννήτριας συνταγών

Τώρα που εξετάσαμε ένα σενάριο, ας γράψουμε κώδικα που να ταιριάζει με το σενάριο που παρουσιάστηκε. Για να το κάνετε αυτό, ακολουθήστε τα παρακάτω βήματα:

1. Χρησιμοποιήστε το υπάρχον αρχείο ως αφετηρία
1. Δημιουργήστε μια μεταβλητή `prompt` και αλλάξτε το δείγμα κώδικα όπως παρακάτω:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Ωραία, ας δούμε πώς μπορούμε να το βελτιώσουμε. Για να το βελτιώσουμε, θέλουμε να βεβαιωθούμε ότι ο κώδικας είναι ευέλικτος, ώστε τα υλικά και ο αριθμός των συνταγών να μπορούν να βελτιωθούν και να αλλάξουν.


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

token = os.environ["GITHUB_TOKEN"]
endpoint = "https://models.inference.ai.azure.com"

model_name = "gpt-4o"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````
