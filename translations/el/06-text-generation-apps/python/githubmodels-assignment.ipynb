# Δημιουργήστε εφαρμογές παραγωγής κειμένου

Μέχρι τώρα, μέσα από αυτό το πρόγραμμα σπουδών έχετε δει ότι υπάρχουν βασικές έννοιες όπως οι προτροπές και ακόμη και μια ολόκληρη πειθαρχία που ονομάζεται "μηχανική προτροπών". Πολλά εργαλεία με τα οποία μπορείτε να αλληλεπιδράσετε, όπως το ChatGPT, το Office 365, το Microsoft Power Platform και άλλα, σας υποστηρίζουν χρησιμοποιώντας προτροπές για να επιτύχετε κάτι.

Για να προσθέσετε αυτήν την εμπειρία σε μια εφαρμογή, πρέπει να κατανοήσετε έννοιες όπως οι προτροπές, οι ολοκληρώσεις και να επιλέξετε μια βιβλιοθήκη για να εργαστείτε. Αυτό ακριβώς θα μάθετε σε αυτό το κεφάλαιο.

## Εισαγωγή

Σε αυτό το κεφάλαιο, θα:

- Μάθετε για τη βιβλιοθήκη openai και τις βασικές της έννοιες.
- Δημιουργήσετε μια εφαρμογή παραγωγής κειμένου χρησιμοποιώντας το openai.
- Κατανοήσετε πώς να χρησιμοποιείτε έννοιες όπως προτροπή, θερμοκρασία και tokens για να δημιουργήσετε μια εφαρμογή παραγωγής κειμένου.

## Στόχοι μάθησης

Στο τέλος αυτού του μαθήματος, θα μπορείτε να:

- Εξηγήσετε τι είναι μια εφαρμογή παραγωγής κειμένου.
- Δημιουργήσετε μια εφαρμογή παραγωγής κειμένου χρησιμοποιώντας το openai.
- Διαμορφώσετε την εφαρμογή σας να χρησιμοποιεί περισσότερα ή λιγότερα tokens και να αλλάζετε τη θερμοκρασία, για ποικιλία στην έξοδο.

## Τι είναι μια εφαρμογή παραγωγής κειμένου;

Κανονικά, όταν δημιουργείτε μια εφαρμογή, αυτή έχει κάποιο είδος διεπαφής, όπως η ακόλουθη:

- Βασισμένη σε εντολές. Οι εφαρμογές κονσόλας είναι τυπικές εφαρμογές όπου πληκτρολογείτε μια εντολή και αυτή εκτελεί μια εργασία. Για παράδειγμα, το `git` είναι μια εφαρμογή βασισμένη σε εντολές.
- Διεπαφή χρήστη (UI). Μερικές εφαρμογές έχουν γραφικές διεπαφές χρήστη (GUIs), όπου κάνετε κλικ σε κουμπιά, εισάγετε κείμενο, επιλέγετε επιλογές και άλλα.

### Οι εφαρμογές κονσόλας και UI έχουν περιορισμούς

Συγκρίνετε το με μια εφαρμογή βασισμένη σε εντολές όπου πληκτρολογείτε μια εντολή:

- **Έχει περιορισμούς**. Δεν μπορείτε απλώς να πληκτρολογήσετε οποιαδήποτε εντολή, μόνο αυτές που υποστηρίζει η εφαρμογή.
- **Ειδικότητα στη γλώσσα**. Μερικές εφαρμογές υποστηρίζουν πολλές γλώσσες, αλλά κατά προεπιλογή, η εφαρμογή είναι κατασκευασμένη για μια συγκεκριμένη γλώσσα, ακόμη και αν μπορείτε να προσθέσετε υποστήριξη για περισσότερες γλώσσες.

### Οφέλη των εφαρμογών παραγωγής κειμένου

Πώς διαφέρει λοιπόν μια εφαρμογή παραγωγής κειμένου;

Σε μια εφαρμογή παραγωγής κειμένου, έχετε μεγαλύτερη ευελιξία, δεν περιορίζεστε σε ένα σύνολο εντολών ή σε μια συγκεκριμένη γλώσσα εισόδου. Αντίθετα, μπορείτε να χρησιμοποιείτε φυσική γλώσσα για να αλληλεπιδράτε με την εφαρμογή. Ένα επιπλέον όφελος είναι ότι, επειδή ήδη αλληλεπιδράτε με μια πηγή δεδομένων που έχει εκπαιδευτεί σε ένα τεράστιο σώμα πληροφοριών, ενώ μια παραδοσιακή εφαρμογή μπορεί να περιορίζεται σε αυτά που υπάρχουν σε μια βάση δεδομένων.

### Τι μπορώ να κατασκευάσω με μια εφαρμογή παραγωγής κειμένου;

Υπάρχουν πολλά πράγματα που μπορείτε να δημιουργήσετε. Για παράδειγμα:

- **Ένα chatbot**. Ένα chatbot που απαντά σε ερωτήσεις σχετικά με θέματα, όπως η εταιρεία σας και τα προϊόντα της, θα μπορούσε να είναι μια καλή επιλογή.
- **Βοηθός**. Τα LLM είναι υπέροχα σε πράγματα όπως η σύνοψη κειμένου, η εξαγωγή πληροφοριών από κείμενο, η παραγωγή κειμένου όπως βιογραφικά και άλλα.
- **Βοηθός κώδικα**. Ανάλογα με το μοντέλο γλώσσας που χρησιμοποιείτε, μπορείτε να δημιουργήσετε έναν βοηθό κώδικα που σας βοηθά να γράψετε κώδικα. Για παράδειγμα, μπορείτε να χρησιμοποιήσετε ένα προϊόν όπως το GitHub Copilot καθώς και το ChatGPT για να σας βοηθήσουν να γράψετε κώδικα.

## Πώς μπορώ να ξεκινήσω;

Λοιπόν, πρέπει να βρείτε έναν τρόπο να ενσωματωθείτε με ένα LLM, που συνήθως περιλαμβάνει τις ακόλουθες δύο προσεγγίσεις:

- Χρήση API. Εδώ κατασκευάζετε αιτήσεις web με την προτροπή σας και λαμβάνετε παραγόμενο κείμενο σε απάντηση.
- Χρήση βιβλιοθήκης. Οι βιβλιοθήκες βοηθούν να ενσωματώσετε τις κλήσεις API και να τις κάνετε πιο εύχρηστες.

## Βιβλιοθήκες/SDKs

Υπάρχουν μερικές γνωστές βιβλιοθήκες για εργασία με LLM όπως:

## Παράγωγή κειμένου με το ChatCompletions

Ο τρόπος για να παραγάγετε κείμενο είναι να χρησιμοποιήσετε την κλάση `ChatCompletionsClient`.
Στο `samples/python/azure_ai_inference/basic.py`, στην ενότητα απόκρισης του κώδικα, ενημερώστε τον κώδικα του ρόλου χρήστη αλλάζοντας την παράμετρο content ως εξής:

```python

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": "Complete the following: Once upon a time there was a",
        },
    ],
    model=model_name,
    # Προαιρετικές παράμετροι
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

```

Τρέξτε το ενημερωμένο αρχείο για να δείτε το αποτέλεσμα


## Διαφορετικοί τύποι προτροπών, για διαφορετικά πράγματα

Τώρα έχετε δει πώς να δημιουργήσετε κείμενο χρησιμοποιώντας μια προτροπή. Έχετε ακόμη ένα πρόγραμμα σε λειτουργία που μπορείτε να τροποποιήσετε και να αλλάξετε για να δημιουργήσετε διαφορετικούς τύπους κειμένου. 

Οι προτροπές μπορούν να χρησιμοποιηθούν για κάθε είδους εργασίες. Για παράδειγμα:

- **Δημιουργία ενός τύπου κειμένου**. Για παράδειγμα, μπορείτε να δημιουργήσετε ένα ποίημα, ερωτήσεις για ένα κουίζ κ.ά.
- **Αναζήτηση πληροφοριών**. Μπορείτε να χρησιμοποιήσετε προτροπές για να αναζητήσετε πληροφορίες όπως το παρακάτω παράδειγμα «Τι σημαίνει CORS στην ανάπτυξη ιστού;».
- **Δημιουργία κώδικα**. Μπορείτε να χρησιμοποιήσετε προτροπές για να δημιουργήσετε κώδικα, για παράδειγμα να αναπτύξετε μια κανονική έκφραση για την επαλήθευση email ή γιατί όχι να δημιουργήσετε ένα ολόκληρο πρόγραμμα, όπως μια web εφαρμογή;

## Άσκηση - δημιουργήστε έναν γεννήτρια συνταγών

Τώρα που έχουμε παίξει ένα σενάριο, ας γράψουμε κώδικα που να ταιριάζει με το επιδεικνυόμενο σενάριο. Για να το κάνετε αυτό, ακολουθήστε τα παρακάτω βήματα:

````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

Τέλεια, ας δούμε πώς μπορούμε να βελτιώσουμε τα πράγματα. Για να βελτιώσουμε τα πράγματα, θέλουμε να διασφαλίσουμε ότι ο κώδικας είναι ευέλικτος, ώστε τα υλικά και ο αριθμός των συνταγών να μπορούν να βελτιωθούν και να αλλάξουν.


1. Ας αλλάξουμε τον κώδικα με τον εξής τρόπο:


````python
import os
from azure.ai.inference import ChatCompletionsClient
from azure.ai.inference.models import SystemMessage, UserMessage
from azure.core.credentials import AzureKeyCredential

# Get these from your Microsoft Foundry project's "Overview" page
token = os.environ["AZURE_INFERENCE_CREDENTIAL"]
endpoint = os.environ["AZURE_INFERENCE_ENDPOINT"]

model_name = "gpt-4o-mini"

client = ChatCompletionsClient(
    endpoint=endpoint,
    credential=AzureKeyCredential(token),
)

no_recipes = input("No of recipes (for example, 5): ")

ingredients = input("List of ingredients (for example, chicken, potatoes, and carrots): ")

# interpolate the number of recipes into the prompt an ingredients
prompt = f"Show me {no_recipes} recipes for a dish with the following ingredients: {ingredients}. Per recipe, list all the ingredients used"

response = client.complete(
    messages=[
        {
            "role": "system",
            "content": "You are a helpful assistant.",
        },
        {
            "role": "user",
            "content": prompt,
        },
    ],
    model=model_name,
    # Optional parameters
    temperature=1.,
    max_tokens=1000,
    top_p=1.    
)

print(response.choices[0].message.content)
````

- **Πειραματισμός με τη θερμοκρασία**. Η θερμοκρασία είναι κάτι που δεν έχουμε αναφέρει μέχρι τώρα αλλά αποτελεί σημαντικό πλαίσιο για το πώς λειτουργεί το πρόγραμμα μας.
