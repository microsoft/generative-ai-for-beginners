## Τι είναι μια εφαρμογή δημιουργίας κειμένου;

Κανονικά, όταν δημιουργείτε μια εφαρμογή, έχει κάποιο είδος διεπαφής όπως η ακόλουθη:

- Βασισμένη σε εντολές. Οι εφαρμογές κονσόλας είναι τυπικές εφαρμογές όπου πληκτρολογείτε μια εντολή και αυτή εκτελεί μια εργασία. Για παράδειγμα, το `git` είναι μια εφαρμογή βασισμένη σε εντολές.
- Διεπαφή χρήστη (UI). Κάποιες εφαρμογές έχουν γραφικές διεπαφές χρήστη (GUI) όπου πατάτε κουμπιά, εισάγετε κείμενο, επιλέγετε επιλογές και άλλα.

### Οι εφαρμογές κονσόλας και UI έχουν περιορισμούς

Συγκρίνετέ το με μια εφαρμογή βασισμένη σε εντολή όπου πληκτρολογείτε μια εντολή: 

- **Έχει περιορισμούς**. Δεν μπορείτε απλά να πληκτρολογήσετε οποιαδήποτε εντολή, μόνο αυτές που υποστηρίζει η εφαρμογή.
- **Ειδικό για γλώσσες**. Κάποιες εφαρμογές υποστηρίζουν πολλές γλώσσες, αλλά από προεπιλογή η εφαρμογή είναι φτιαγμένη για συγκεκριμένη γλώσσα, ακόμα και αν μπορείτε να προσθέσετε επιπλέον υποστήριξη γλωσσών.

## Δημιουργία κειμένου

Ο τρόπος για να δημιουργήσετε κείμενο είναι να χρησιμοποιήσετε τη μέθοδο `responses.create`. Να ένα παράδειγμα:

```python
prompt = "Complete the following: Once upon a time there was a"

response = client.responses.create(model=deployment, input=prompt, store=False)
print(response.output_text)
```

Στον παραπάνω κώδικα, δημιουργούμε ένα αντικείμενο απόκρισης και περνάμε το μοντέλο που θέλουμε να χρησιμοποιήσουμε και το prompt. Στη συνέχεια εκτυπώνουμε το παραγόμενο κείμενο.

### Απαντήσεις διαλόγου

Το Responses API είναι ιδανικό τόσο για μονής περιστροφής δημιουργία κειμένου όσο και για πολύστροφους chatbots - απλά παρέχετε περισσότερα μηνύματα στη λίστα `input` για να χτίσετε μια συνομιλία:

```python
client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

response = client.responses.create(model=deployment, input=[{"role": "user", "content": "Hello world"}], store=False)
print(response.output_text)
```

Θα βρείτε περισσότερες πληροφορίες για αυτή τη λειτουργικότητα σε κεφάλαιο που έρχεται.

## Άσκηση - δημιουργήστε έναν γεννήτρια συνταγών

Τώρα που έχουμε παίξει ένα σενάριο, ας γράψουμε κώδικα που να ταιριάζει με το παρουσιασμένο σενάριο. Για να το κάνετε αυτό, ακολουθήστε τα εξής βήματα:

1. Χρησιμοποιήστε το υπάρχον αρχείο *app.py* ως σημείο εκκίνησης
1. Εντοπίστε τη μεταβλητή `prompt` και αλλάξτε τον κώδικά της ως εξής:


````python
import os
from openai import OpenAI
from dotenv import load_dotenv

# load environment variables from .env file
load_dotenv()

client = OpenAI(
  api_key=os.environ['AZURE_OPENAI_API_KEY'],
  base_url=f"{os.environ['AZURE_OPENAI_ENDPOINT'].rstrip('/')}/openai/v1/",
  )

deployment = os.environ['AZURE_OPENAI_DEPLOYMENT']

prompt = "Show me 5 recipes for a dish with the following ingredients: chicken, potatoes, and carrots. Per recipe, list all the ingredients used"

# make a request using the Responses API
response = client.responses.create(model=deployment, input=prompt, max_output_tokens=600, store=False)

# print response
print(response.output_text)
````

Αν τώρα εκτελέσετε τον κώδικα, θα πρέπει να δείτε μια έξοδο παρόμοια με:

- **A word on token length**. We should consider how many tokens we need to generate the text we want. Tokens cost money, so where possible, we should try to be economical with the number of tokens we use. For example, can we phrase the prompt so that we can use less tokens?

- **Experimenting with temperature**. Temperature is something we haven't mentioned so far but is an important context for how our program performs.
